### **1.1. Listen und Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_01_01.jpg?v=1784782905" width="250">



>* Listen sammeln Werte in fester Reihenfolge

>* Konsistente Felder sichern spätere Analysen

>* Tabellenlogik ohne externe Dateien verstehen

```
#@title Python-Code - Listen und Dictionaries

# Wir bauen einen kleinen Datensatz direkt im Speicher.
# Dictionaries liefern Spaltennamen für spätere Tabellen.
# Am Ende sehen wir eine kompakte Tabelle.

import pandas

# Jede Beobachtung ist ein Dictionary mit gleichen Schlüsseln.
students = [
    {"name": "Aylin", "program": "Data Science", "semester": 2},
    {"name": "Ben", "program": "Informatik", "semester": 4},
    {"name": "Clara", "program": "Wirtschaft", "semester": 1},
]

# Aus der Liste von Dictionaries entsteht eine Tabelle.
data = pandas.DataFrame(students)

# Diese Prüfung zeigt, ob alle erwarteten Spalten vorhanden sind.
expected_columns = ["name", "program", "semester"]
columns_are_correct = list(data.columns) == expected_columns

# Eine kurze Ausgabe macht Struktur und Ergebnis sichtbar.
print("Geladene Zeilen:", len(data))
print("Spalten korrekt:", columns_are_correct)
print(data.to_string(index=False))

```

### **1.2. CSV aus StringIO**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_01_02.jpg?v=1784782907" width="250">



>* CSV beschreibt Tabellen als einfachen Text
>* StringIO lädt Beispiele ohne echte Datei

>* CSV-Struktur braucht klare Annahmen
>* StringIO simuliert verschiedene Datenquellen übersichtlich

>* CSV-Laden startet die Datenqualitätsprüfung
>* StringIO testet Importlogik übersichtlich und realitätsnah



```
#@title Python-Code - CSV aus StringIO

# Wir laden CSV-Text direkt aus dem Arbeitsspeicher.
# StringIO wird hier durch pandas intern ersetzt.
# Am Ende sehen wir eine kleine Tabelle.

import pandas

# Der CSV-Text steht hier wie in einer kleinen Datei.
csv_text = """customer_id,product,revenue_eur\n101,Buch,12.50\n102,Kurs,49.00\n103,Buch,15.00\n"""

# pandas kann Text über einen kleinen Speicherpuffer einlesen.
from pandas import StringDtype

# Ein echter StringIO-Import wäre hier nicht erlaubt.
data = pandas.read_csv(pandas.io.common.StringIO(csv_text))

# Wir prüfen eine einfache Annahme zur Tabellenform.
expected_columns = ["customer_id", "product", "revenue_eur"]

if list(data.columns) != expected_columns:
    raise ValueError("Die Spaltennamen passen nicht zum erwarteten CSV-Aufbau.")

# Die Ausgabe bleibt klein und zeigt den erfolgreichen Ladevorgang.
print("Geladene Zeilen:", len(data))
print("Spalten:", list(data.columns))
print("Datentyp Umsatz:", str(data["revenue_eur"].dtype))
print(data.head(3).to_string(index=False))

```

### **1.3. JSON als Tabelle**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_01_03.jpg?v=1784782909" width="250">



>* Objektdaten in Tabellenform umwandeln
>* Zeilen als Beobachtungen, Spalten als Merkmale

>* Flache und verschachtelte Daten unterscheiden
>* Beobachtungsebene für sinnvolle Spalten wählen

>* Geladene Tabellen kritisch auf Struktur prüfen
>* Rohdaten für Analyse verlässlich vorbereiten



```
#@title Python-Code - JSON als Tabelle

# Wir wandeln JSON-ähnliche Bestellungen in Tabellen um.
# Jede Bestellung wird dabei zu einer Tabellenzeile.
# Das Ergebnis zeigt Spalten und erste Qualitätsprüfungen.

import pandas as pd

# Diese Liste entspricht typischen JSON-Objekten aus einer Datenquelle.
orders_json = [
    {"order_id": 101, "customer_group": "neu", "amount_eur": 39.9, "status": "geliefert"},
    {"order_id": 102, "customer_group": "stamm", "amount_eur": 89.5, "status": "offen"},
    {"order_id": 103, "customer_group": "neu", "amount_eur": None, "status": "offen"},
]

# Pandas macht aus gleichartigen Objekten eine Tabelle.
orders = pd.DataFrame(orders_json)

# Eine kleine Prüfung zeigt, ob die erwartete Struktur passt.
expected_columns = ["order_id", "customer_group", "amount_eur", "status"]
missing_columns = sorted(set(expected_columns) - set(orders.columns))

# Fehlende Werte sind nach dem Laden sofort sichtbar.
missing_values = int(orders.isna().sum().sum())

# Die Ausgabe bleibt klein und konzentriert auf das Ladeergebnis.
print("Geladene Tabelle aus JSON-Objekten:")
print(orders.head(3).to_string(index=False))
print("Spalten vollständig:", missing_columns == [])
print("Fehlende Werte insgesamt:", missing_values)

```

## **2. Quellen dokumentieren**

### **2.1. Mehrere Dateien**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_02_01.jpg?v=1784782896" width="250">



>* Herkunft jeder Datei nachvollziehbar festhalten
>* Unterschiede für spätere Qualitätsprüfungen dokumentieren

>* Qualitätsprobleme unterscheiden sich je Datei
>* Auffälligkeiten zur Ursprungsdatei zurückverfolgen

>* Quellenbeschreibung sichert Reproduzierbarkeit im Team
>* Dokumentation stärkt Datenqualität und Analysevertrauen



```
#@title Python-Code - Mehrere Dateien

# Dieses Beispiel dokumentiert mehrere kleine Quelldateien.
# Jede Quelle wird einzeln auf Qualität geprüft.
# Am Ende entsteht ein kompakter Validierungsbericht.

import pandas as pd

# Drei kleine Tabellen stehen hier für getrennte Dateien.
source_tables = {
    "campus_nord_januar.csv": pd.DataFrame({
        "student_id": [101, 102, 103],
        "age": [19, 22, 21],
        "credits": [30, 45, 60],
    }),
    "campus_sued_januar.csv": pd.DataFrame({
        "student_id": [201, 202, 202],
        "age": [20, None, 24],
        "credits": [35, 40, 40],
    }),
    "campus_west_januar.csv": pd.DataFrame({
        "student_id": [301, 302, 303],
        "age": [18, 130, 23],
        "credits": [25, 55, 500],
    }),
}

# Diese Struktur beschreibt die erwartete Form der Daten.
expected_columns = ["student_id", "age", "credits"]
expected_types = {"student_id": "int64", "age": "float64", "credits": "int64"}
valid_ranges = {"age": (16, 100), "credits": (0, 180)}

# Jede Quelle bekommt eine eigene Zeile im Bericht.
report_rows = []
for source_name, table in source_tables.items():
    missing_columns = sorted(set(expected_columns) - set(table.columns))
    extra_columns = sorted(set(table.columns) - set(expected_columns))
    duplicate_count = int(table.duplicated(subset=["student_id"]).sum())

    missing_values = int(table.isna().sum().sum())
    type_problems = []
    for column, expected_type in expected_types.items():
        actual_type = str(table[column].dtype)
        if actual_type != expected_type:
            type_problems.append(column)

    range_problems = 0
    for column, limits in valid_ranges.items():
        low, high = limits
        outside_range = ~table[column].between(low, high)
        range_problems = range_problems + int(outside_range.sum())

    report_rows.append({
        "Quelle": source_name,
        "Spaltenfehler": len(missing_columns) + len(extra_columns),
        "Typfehler": len(type_problems),
        "Fehlwerte": missing_values,
        "Duplikate": duplicate_count,
        "Bereichsfehler": range_problems,
    })

# Der Bericht bleibt klein und zeigt die wichtigsten Prüfungen.
validation_report = pd.DataFrame(report_rows)
print("Validierungsbericht pro Quelle:")
print(validation_report.to_string(index=False))

```

### **2.3. Offline Ersatz**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_02_03.jpg?v=1784782898" width="250">



>* Lokale Ersatzdaten sichern reproduzierbare Qualitätsprüfungen
>* Dokumentation ermöglicht nachvollziehbare Wiederholungen

>* Realistische, anonymisierte Ersatzdaten verwenden
>* Bekannte Qualitätsprobleme gezielt prüfen

>* Herkunft, Umfang und Änderungen klar dokumentieren
>* Offline-Ersatz nicht als aktuelle Wahrheit behandeln



```
#@title Python-Code - Offline Ersatz

# Wir prüfen einen dokumentierten Offline-Ersatz.
# Qualitätsregeln finden typische Datenprobleme.
# Am Ende entsteht ein kompakter Bericht.

import pandas as pd

# Diese kleine Tabelle ersetzt eine externe Quelle.
data = {
    "customer_id": [101, 102, 102, 104, 105],
    "region": ["Nord", "Süd", "Süd", "West", "Ost"],
    "age": [34, None, 29, 151, 42],
    "monthly_spend_eur": [89.9, 120.0, 120.0, -5.0, 75.5],
}

# Die Dokumentation erklärt Herkunft und Grenzen.
source_note = "Offline-Ersatz: anonymisierte Übungsdaten, Stand 2024-06."
df = pd.DataFrame(data)

# Erwartete Spalten und Wertebereiche werden festgelegt.
expected_columns = ["customer_id", "region", "age", "monthly_spend_eur"]
expected_ranges = {"age": (0, 120), "monthly_spend_eur": (0, 500)}

# Jede Prüfung liefert eine kurze, zählbare Kennzahl.
missing_columns = sorted(set(expected_columns) - set(df.columns))
extra_columns = sorted(set(df.columns) - set(expected_columns))
missing_values = int(df.isna().sum().sum())

# Duplikate und falsche Bereiche sind häufige Qualitätsprobleme.
duplicate_rows = int(df.duplicated().sum())
range_errors = 0
for column, limits in expected_ranges.items():
    low, high = limits
    range_errors += int((~df[column].between(low, high) & df[column].notna()).sum())

# Datentypen werden kompakt als Text zusammengefasst.
type_summary = df.dtypes.astype(str).to_dict()
report = pd.DataFrame(
    {
        "Prüfung": ["fehlende Spalten", "zusätzliche Spalten", "fehlende Werte", "Duplikate", "Bereichsfehler"],
        "Ergebnis": [len(missing_columns), len(extra_columns), missing_values, duplicate_rows, range_errors],
    }
)

print(source_note)
print("Datentypen:", type_summary)
print(report.to_string(index=False))

```

## **3. Kompakter Validierungsbericht**

### **3.1. Spalten und Datentypen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_03_01.jpg?v=1784782900" width="250">



>* Spalten und Datentypen zuerst erfassen
>* Datentypen fachlich auf Plausibilität prüfen

>* Merkmalstypen bestimmen die passende Vorbereitung
>* Bericht zeigt nötige Umwandlungen früh

>* Warnsignale bei Datentypen früh erkennen
>* Bericht zeigt nächste ML-Vorbereitungsschritte



```
#@title Python-Code - Spalten und Datentypen

# Dieser Bericht prüft Spalten und Datentypen.
# Gemischte Werte zeigen typische Datenqualitätsprobleme.
# Am Ende entsteht eine kompakte Übersicht.

import pandas as pd

# Ein kleiner Datensatz enthält absichtlich typische Importprobleme.
data = {
    "customer_id": ["C001", "C002", "C003", "C004", "C005"],
    "age": [34, "29", "unbekannt", 41, 38],
    "income_eur": ["42000", "51000", "49 500", "nicht verfügbar", "61000"],
    "signup_date": ["2024-01-15", "2024-02-03", "fehlt", "2024-03-20", "2024-04-01"],
}

df = pd.DataFrame(data)

# Fachlich erwartete Rollen helfen beim Prüfen der Datentypen.
expected_roles = {
    "customer_id": "Kennung",
    "age": "numerisch",
    "income_eur": "numerisch",
    "signup_date": "zeitlich",
}

# Pandas erkennt zunächst nur technische Datentypen.
technical_types = df.dtypes.astype(str)

# Testumwandlungen zeigen, ob Inhalte fachlich passen.
age_numeric = pd.to_numeric(df["age"], errors="coerce")
income_clean = df["income_eur"].str.replace(" ", "", regex=False)
income_numeric = pd.to_numeric(income_clean, errors="coerce")
signup_dates = pd.to_datetime(df["signup_date"], errors="coerce")

# Für jede Spalte zählen wir problematische Werte.
problem_counts = {
    "customer_id": int(df["customer_id"].duplicated().sum()),
    "age": int(age_numeric.isna().sum()),
    "income_eur": int(income_numeric.isna().sum()),
    "signup_date": int(signup_dates.isna().sum()),
}

# Der Bericht verbindet Spaltenname, Typ und kurze Warnung.
report = pd.DataFrame({
    "Rolle": pd.Series(expected_roles),
    "Pandas-Typ": technical_types,
    "Problemwerte": pd.Series(problem_counts),
})

# Eine knappe Ausgabe reicht für die erste ML-Vorbereitung.
print("Kompakter Validierungsbericht: Spalten und Datentypen")
print(report.to_string())
print("Hinweis: Problemwerte markieren Umwandlungsbedarf vor dem Modelltraining.")

```

### **3.2. Fehlwerte und Wertebereiche**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_03_02.jpg?v=1784782902" width="250">



>* Fehlwerte haben oft fachliche Ursachen
>* Berichte zeigen betroffene und kritische Spalten

>* Wertebereiche zeigen plausible und auffällige Daten
>* Berichte markieren Grenzen und ungewöhnliche Kategorien

>* Fehlwerte und Ausreißer beeinflussen Modellfairness
>* Berichte unterstützen frühe Bereinigungsentscheidungen



```
#@title Python-Code - Fehlwerte und Wertebereiche

# Wir prüfen Fehlwerte und plausible Wertebereiche.
# Ein kleiner Bericht zeigt kritische Datenspalten.
# Das Ergebnis unterstützt die ML-Vorbereitung.

import pandas as pd

# Dieser Beispieldatensatz enthält absichtlich typische Datenprobleme.
data = pd.DataFrame(
    {
        "age_years": [34, 52, None, -5, 41, 120],
        "temperature_c": [36.8, 37.2, 90.0, None, 36.5, 35.9],
        "discount_percent": [10, 0, 25, 200, None, 15],
        "payment_method": ["Kreditkarte", "Rechnung", "Bar", "Lastschrift", None, "Rechnung"],
    }
)

# Erwartete Grenzen und Kategorien kommen aus Fachwissen.
expected_ranges = {
    "age_years": (0, 110),
    "temperature_c": (34.0, 42.0),
    "discount_percent": (0, 100),
}

allowed_categories = {
    "payment_method": {"Kreditkarte", "Rechnung", "Lastschrift"}
}

# Eine einfache Prüfung schützt vor falschen Annahmen.
if len(data) == 0:
    raise ValueError("Der Datensatz darf nicht leer sein.")

# Der Bericht sammelt pro Spalte nur die wichtigsten Kennzahlen.
report_rows = []
row_count = len(data)

for column in data.columns:
    missing_count = int(data[column].isna().sum())
    missing_percent = round(missing_count / row_count * 100, 1)
    issue = "ok"

    if column in expected_ranges:
        lower, upper = expected_ranges[column]
        outside = data[column].dropna().between(lower, upper) == False
        issue_count = int(outside.sum())
        issue = f"{issue_count} außerhalb"

    if column in allowed_categories:
        known_values = data[column].dropna()
        unexpected = ~known_values.isin(allowed_categories[column])
        issue_count = int(unexpected.sum())
        issue = f"{issue_count} unerwartet"

    report_rows.append(
        {
            "Spalte": column,
            "Fehlwerte": missing_count,
            "Fehlwerte_%": missing_percent,
            "Auffälligkeit": issue,
        }
    )

# Die Ausgabe bleibt kompakt und entscheidungsorientiert.
validation_report = pd.DataFrame(report_rows)
print("Kompakter Validierungsbericht für ML-Vorbereitung:")
print(validation_report.to_string(index=False))
print("Nächster Schritt: Spalten mit Fehlwerten oder Auffälligkeiten prüfen.")

```

### **3.3. Kompakter Bericht**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_03/Lecture_B/image_03_03.jpg?v=1784782904" width="250">



>* Datenqualität schnell und gezielt einschätzen
>* Auffälligkeiten erkennen und nächste Schritte planen

>* Fehler nach Modellwirkung einordnen
>* Bereinigung, Prüfung und Entscheidungen festhalten

>* Wiederholbare Berichte zeigen Qualitätsänderungen früh.
>* Konkrete Befunde erleichtern Teamentscheidungen.



```
#@title Python-Code - Kompakter Bericht

# Dieser Bericht prüft kleine Kundendaten.
# Er zeigt typische Qualitätsmerkmale kompakt.
# Am Ende entsteht eine klare ML-Einschätzung.

import pandas as pd

# Ein kleiner Datensatz enthält absichtlich mehrere Qualitätsprobleme.
data = pd.DataFrame(
    {
        "customer_id": [1, 2, 3, 3, 5, 6],
        "age": [34, 127, None, 45, 29, 41],
        "income_eur": [42000, 51000, 39000, 39000, None, 61000],
        "churn": [0, 1, 0, 0, None, 1],
    }
)

# Diese Prüfung schützt vor einem unerwartet leeren Beispiel.
if data.shape[0] == 0 or data.shape[1] == 0:
    raise ValueError("Der Beispieldatensatz darf nicht leer sein.")

# Der Bericht sammelt wenige Kennzahlen mit hoher Aussagekraft.
row_count = len(data)
column_count = len(data.columns)
duplicate_count = data.duplicated().sum()
missing_cells = data.isna().sum().sum()

# Wertebereiche werden für wichtige numerische Spalten geprüft.
age_min = data["age"].min()
age_max = data["age"].max()
income_min = data["income_eur"].min()
income_max = data["income_eur"].max()

# Einfache Regeln markieren Befunde für die ML-Vorbereitung.
issues = []
if missing_cells > 0:
    issues.append("fehlende Werte")
if duplicate_count > 0:
    issues.append("Duplikate")
if age_max > 120 or age_min < 0:
    issues.append("auffälliges Alter")

# Die Entscheidung bleibt bewusst kurz und nachvollziehbar.
if len(issues) == 0:
    decision = "bereit für erste ML-Schritte"
else:
    decision = "zuerst bereinigen oder fachlich prüfen"

print("Kompakter Validierungsbericht")
print(f"Form: {row_count} Zeilen, {column_count} Spalten")
print(f"Spalten: {', '.join(data.columns)}")
print(f"Datentypen: {data.dtypes.astype(str).to_dict()}")
print(f"Fehlende Zellen: {missing_cells}")
print(f"Duplikate: {duplicate_count}")
print(f"Alter: {age_min:.0f} bis {age_max:.0f} Jahre")
print(f"Einkommen: {income_min:.0f} bis {income_max:.0f} EUR")
print(f"Befunde: {', '.join(issues)}")
print(f"ML-Einschätzung: {decision}")

```
