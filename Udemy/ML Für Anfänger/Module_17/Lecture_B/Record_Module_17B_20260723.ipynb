## **3. Reproduzierbar speichern**

### **3.1. Metriken visualisieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_17/Lecture_B/image_03_01.jpg?v=1784814115" width="250">



>* Trainingsverlauf begründet gespeicherte Modellzustände
>* Kurven zeigen Lernen, Stagnation oder Überanpassung

>* state_dict speichert keine Trainingsgeschichte
>* Metriken zeigen Überanpassung und Risiken

>* Metriken begründen und prüfen gespeicherte Modelle
>* Visualisierungen fördern Teamverständnis und Wiederholbarkeit



```
#@title Python-Code - Metriken visualisieren

# Wir visualisieren Metriken eines gespeicherten Modellzustands.
# Die Kurven zeigen Training und Validierung.
# Der beste Speicherzeitpunkt wird sichtbar markiert.

import numpy as np
import matplotlib.pyplot as plt

# Diese Werte stehen für protokollierte Trainingsmetriken.
epochs = np.arange(1, 11)
train_loss = np.array([0.92, 0.74, 0.61, 0.50, 0.42, 0.36, 0.31, 0.27, 0.24, 0.21])
valid_loss = np.array([0.95, 0.78, 0.64, 0.55, 0.49, 0.47, 0.48, 0.51, 0.55, 0.60])

# Der kleinste Validierungsverlust bestimmt den gespeicherten Zustand.
best_index = int(np.argmin(valid_loss))
best_epoch = int(epochs[best_index])
best_valid_loss = float(valid_loss[best_index])

# Eine einfache Prüfung schützt vor falsch protokollierten Kurven.
if len(epochs) != len(train_loss) or len(epochs) != len(valid_loss):
    raise ValueError("Die Metriklisten müssen gleich lang sein.")

# Diese Ausgabe verbindet die Kurve mit state_dict.
print(f"Gespeichertes state_dict: Epoche {best_epoch}")
print(f"Bester Validierungsverlust: {best_valid_loss:.2f}")
print("Nach dem Laden sollten Vorhersagen zu diesem Zustand passen.")

# Die Visualisierung macht den Speicherzeitpunkt nachvollziehbar.
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(epochs, train_loss, marker="o", label="Trainingsverlust")
ax.plot(epochs, valid_loss, marker="o", label="Validierungsverlust")

# Die Markierung zeigt den empfohlenen gespeicherten Zustand.
ax.scatter(best_epoch, best_valid_loss, color="red", s=90, label="gespeichert")
ax.set_title("Metriken zum gespeicherten Modellzustand")
ax.set_xlabel("Epoche")
ax.set_ylabel("Verlust")

# Eine Legende hilft beim Vergleich der Kurven.
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()

```
