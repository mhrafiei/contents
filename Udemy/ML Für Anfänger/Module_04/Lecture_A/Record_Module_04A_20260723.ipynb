### **1.1. Erste Tabellenübersicht**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_01_01.jpg?v=1784783820" width="250">



>* Zeilen, Spalten und Werte zuerst verstehen

>* Datentypen und Wertebereiche bewusst prüfen
>* Auffälligkeiten als Plausibilitätshinweise erkennen

>* Datenstruktur kritisch auf Realität prüfen
>* Granularität und Unsicherheiten früh erkennen



```
#@title Python-Code - Erste Tabellenübersicht

# Diese Übung zeigt eine erste Tabellenübersicht.
# Wir prüfen Struktur, Datentypen und Kennzahlen.
# Am Ende entsteht ein kompakter Analyseüberblick.

import pandas as pd

# Ein kleiner Datensatz bleibt übersichtlich und nachvollziehbar.
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "program": ["Informatik", "BWL", "Informatik", "Design", "BWL", "Design"],
    "semester": [1, 3, 2, 5, 4, 2],
    "exam_score": [1.7, 2.3, 1.3, 3.0, 2.0, 5.0],
    "attendance_rate": [0.92, 0.81, 0.97, 0.66, 0.88, 0.40],
}

# Aus dem Wörterbuch entsteht eine Tabelle.
students = pd.DataFrame(data)

# Eine einfache Prüfung schützt vor leeren Tabellen.
if students.shape[0] == 0 or students.shape[1] == 0:
    raise ValueError("Die Tabelle enthält keine auswertbaren Daten.")

# Die Übersicht kombiniert Größe, Spalten und Datentypen.
print("Tabellengröße:", students.shape[0], "Zeilen und", students.shape[1], "Spalten")
print("Spalten:", ", ".join(students.columns))
print("Datentypen:", students.dtypes.astype(str).to_dict())

# Die ersten Zeilen zeigen typische Werte ohne Überforderung.
print("Erste Zeilen:")
print(students.head(3).to_string(index=False))

# Numerische Kennzahlen fassen Wertebereiche kompakt zusammen.
summary = students[["semester", "exam_score", "attendance_rate"]].describe()
summary = summary.loc[["mean", "min", "max"]].round(2)
print("Kennzahlen für numerische Spalten:")
print(summary.to_string())

# Fehlende Werte werden früh sichtbar gemacht.
missing_values = students.isna().sum().to_dict()
print("Fehlende Werte je Spalte:", missing_values)

```

### **1.2. Kennzahlen vergleichen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_01_02.jpg?v=1784783824" width="250">



>* Mittelwert und Median gemeinsam betrachten
>* Ausreißer können Durchschnitte stark verzerren

>* Streuung zeigt Unterschiede trotz gleichem Durchschnitt
>* Spannweite, Quartile und Standardabweichung einordnen

>* Kennzahlen nach Gruppen und Zeit vergleichen
>* Unterschiede vorsichtig im Kontext deuten



```
#@title Python-Code - Kennzahlen vergleichen

# Wir vergleichen Kennzahlen einer kleinen Tabelle.
# Mittelwert und Median zeigen unterschiedliche Perspektiven.
# Ausreißer werden durch den Vergleich sichtbar.

import pandas as pd

# Diese Umsätze enthalten einen ungewöhnlich hohen Ausreißer.
sales = pd.DataFrame(
    {"day": ["Mo", "Di", "Mi", "Do", "Fr"], "sales_eur": [210, 220, 215, 225, 900]}
)

# Eine kurze Prüfung verhindert leere oder falsche Beispiele.
if len(sales) == 0 or "sales_eur" not in sales.columns:
    raise ValueError("Die Tabelle braucht eine Spalte sales_eur.")

# Wir berechnen Lage und Streuung gemeinsam.
summary = pd.DataFrame(
    {"Kennzahl": ["Mittelwert", "Median", "Minimum", "Maximum", "Spannweite"],
     "Wert in Euro": [sales["sales_eur"].mean(), sales["sales_eur"].median(), sales["sales_eur"].min(), sales["sales_eur"].max(), sales["sales_eur"].max() - sales["sales_eur"].min()]}
)

# Gerundete Werte sind für den ersten Überblick leichter lesbar.
summary["Wert in Euro"] = summary["Wert in Euro"].round(1)

# Die Ausgabe zeigt, warum eine Kennzahl allein riskant ist.
print("Tagesumsätze in Euro: 210, 220, 215, 225, 900")
print(summary.to_string(index=False))
print("Der Mittelwert steigt stark, der Median bleibt näher an vier Tagen.")

```

### **1.3. Kategorien zählen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_01_03.jpg?v=1784783822" width="250">



>* Kategorien zeigen die Zusammensetzung des Datensatzes
>* Häufigkeiten machen Vergleiche und Auffälligkeiten sichtbar

>* Anteile zeigen die Bedeutung von Kategorien
>* Zählungen decken uneinheitliche Eingaben auf

>* Kategoriezählungen immer im Kontext deuten
>* Auffälligkeiten als Analysefragen nutzen



```
#@title Python-Code - Kategorien zählen

# Wir zählen Kategorien in einer kleinen Tabelle.
# Absolute Häufigkeiten zeigen die Gruppengrößen.
# Prozentanteile machen die Verteilung leichter vergleichbar.

import pandas as pd

# Diese Beispieldaten enthalten typische Schreibvarianten.
data = pd.DataFrame(
    {
        "payment_method": [
            "Karte", "PayPal", "Karte", "Rechnung", "Karte",
            "paypal", "Rechnung", "Karte ", "PayPal", "Karte",
        ]
    }
)

# Wir prüfen zuerst die Anzahl der Beobachtungen.
row_count = len(data)
if row_count == 0:
    raise ValueError("Die Tabelle enthält keine Zeilen.")

# Roh gezählte Kategorien zeigen auch uneinheitliche Eingaben.
raw_counts = data["payment_method"].value_counts(dropna=False)
print("Roh gezählte Kategorien:")
print(raw_counts.head(5).to_string())

# Vereinheitlichung entfernt Leerzeichen und Großkleinschreibung.
data["payment_clean"] = data["payment_method"].str.strip().str.lower()
clean_counts = data["payment_clean"].value_counts().sort_index()

# Prozentanteile ergänzen die absoluten Häufigkeiten.
summary = pd.DataFrame(
    {
        "anzahl": clean_counts,
        "anteil_prozent": (clean_counts / row_count * 100).round(1),
    }
)

# Die Übersicht verbindet Häufigkeit und Anteil.
print("Bereinigte Übersicht:")
print(summary.to_string())

```

## **2. Diagramme gezielt nutzen**

### **2.1. Verteilungen sichtbar machen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_02_01.jpg?v=1784783818" width="250">



>* Diagramme zeigen Muster hinter Kennzahlen
>* Verteilungen helfen, passende Analysen zu wählen

>* Verteilungsformen zeigen typische Datenmuster
>* Muster verhindern irreführende Durchschnittsdeutungen

>* Diagrammwahl zeigt unterschiedliche Verteilungsaspekte
>* Mehrere Ansichten ermöglichen kritischere Datenanalyse



```
#@title Python-Code - Verteilungen sichtbar machen

# Dieses Beispiel macht Verteilungen mit Histogrammen sichtbar.
# Zwei Gruppen teilen denselben Wertebereich unterschiedlich.
# Das Diagramm zeigt Muster jenseits des Mittelwerts.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wir erzeugen reproduzierbare Prüfungsdaten mit zwei Teilgruppen.
rng = np.random.default_rng(42)
low_scores = rng.normal(loc=58, scale=7, size=90)
high_scores = rng.normal(loc=84, scale=6, size=90)

# Werte werden auf den sinnvollen Bereich begrenzt.
scores = np.concatenate([low_scores, high_scores])
scores = np.clip(scores, 0, 100)
groups = ["Gruppe A"] * 90 + ["Gruppe B"] * 90

# Ein DataFrame macht die Daten übersichtlich nutzbar.
data = pd.DataFrame({"score": scores, "group": groups})
if len(data) != 180:
    raise ValueError("Die Beispieldaten sollten 180 Zeilen enthalten.")

# Kurze Kennzahlen zeigen, was eine Tabelle zusammenfasst.
mean_score = data["score"].mean()
median_score = data["score"].median()
print(f"Mittelwert aller Ergebnisse: {mean_score:.1f} Punkte")
print(f"Median aller Ergebnisse: {median_score:.1f} Punkte")
print("Das Histogramm zeigt zusätzlich die Form der Verteilung.")

# Ein Histogramm macht Häufungen und mögliche Teilgruppen sichtbar.
plt.figure(figsize=(8, 5))
ax = sns.histplot(data=data, x="score", bins=18, color="#4C72B0")

# Beschriftungen helfen bei der fachlichen Interpretation.
ax.axvline(mean_score, color="#C44E52", linewidth=2, label="Mittelwert")
ax.set_title("Verteilung von Prüfungsergebnissen")
ax.set_xlabel("Ergebnis in Punkten")
ax.set_ylabel("Anzahl der Beobachtungen")
ax.legend()

plt.tight_layout()
plt.show()

```

### **2.2. Zusammenhänge und Trends**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_02_02.jpg?v=1784783814" width="250">



>* Streudiagramme zeigen Beziehungen zwischen Merkmalen
>* Punktmuster, Cluster und Ausreißer beachten

>* Liniendiagramme zeigen Entwicklungen über Zeit
>* Zeitauflösung balanciert Detail und Überblick

>* Muster beweisen keine Ursache
>* Gruppen, Skalen und Ausreißer kritisch prüfen



```
#@title Python-Code - Zusammenhänge und Trends

# Dieses Beispiel zeigt Zusammenhänge und Trends.
# Streupunkte machen Muster und Ausreißer sichtbar.
# Eine Trendlinie hilft beim vorsichtigen Interpretieren.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Wir erzeugen kleine, nachvollziehbare Beispieldaten.
rng = np.random.default_rng(42)
area_sqm = np.linspace(35, 120, 45)
noise = rng.normal(0, 120, size=area_sqm.size)

# Die Miete steigt tendenziell mit der Wohnfläche.
rent_eur = 420 + 11 * area_sqm + noise
rent_eur[38] = rent_eur[38] + 650

# Eine Tabelle bündelt die beiden betrachteten Merkmale.
data = pd.DataFrame({"Wohnfläche_m2": area_sqm, "Miete_EUR": rent_eur})

# Eine einfache Prüfung schützt vor leeren oder unpassenden Daten.
if data.shape[0] < 2 or data.shape[1] != 2:
    raise ValueError("Die Beispieldaten brauchen zwei Spalten und mehrere Zeilen.")

# Die Korrelation beschreibt Richtung und Stärke des linearen Musters.
correlation = data["Wohnfläche_m2"].corr(data["Miete_EUR"])
print(f"Korrelation Wohnfläche und Miete: {correlation:.2f}")
print("Wichtig: Korrelation zeigt ein Muster, aber keine Ursache.")

# Eine Trendlinie fasst das sichtbare Muster grob zusammen.
slope, intercept = np.polyfit(data["Wohnfläche_m2"], data["Miete_EUR"], 1)
trend_rent = slope * data["Wohnfläche_m2"] + intercept

# Das Streudiagramm zeigt Punkte, Streuung und Ausreißer gemeinsam.
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(data["Wohnfläche_m2"], data["Miete_EUR"], alpha=0.75, label="Wohnungen")
ax.plot(data["Wohnfläche_m2"], trend_rent, color="red", label="Trendlinie")

ax.set_title("Zusammenhang zwischen Wohnfläche und Miete")
ax.set_xlabel("Wohnfläche in Quadratmetern")
ax.set_ylabel("Monatsmiete in Euro")
ax.legend()

plt.show()

```

### **2.3. Gruppen gezielt vergleichen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_02_03.jpg?v=1784783816" width="250">



>* Gruppen zeigen verborgene Unterschiede.
>* Diagramme machen Streuung und Muster sichtbar.

>* Streuung, Gruppengröße und Verteilung mitbetrachten
>* Klare Visualisierungen zeigen Unsicherheit und Variation

>* Vergleiche brauchen eine klare Fragestellung
>* Unterschiede kritisch und vorsichtig interpretieren



```
#@title Python-Code - Gruppen gezielt vergleichen

# Wir vergleichen Gruppen mit einem Boxplot.
# Streuung ergänzt den reinen Mittelwert.
# Das Diagramm zeigt Unterschiede und Überlappungen.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Feste Zufallszahlen machen das Beispiel reproduzierbar.
rng = np.random.default_rng(42)

# Drei Gruppen erhalten unterschiedliche typische Werte und Streuungen.
groups = ["Filiale A", "Filiale B", "Filiale C"]
means = [78, 78, 70]
spreads = [4, 12, 6]

# Wir erzeugen kleine, künstliche Zufriedenheitswerte pro Gruppe.
rows = []
for group, mean, spread in zip(groups, means, spreads):
    values = rng.normal(loc=mean, scale=spread, size=35)
    values = np.clip(values, 0, 100)
    for value in values:
        rows.append({"group": group, "satisfaction": value})

data = pd.DataFrame(rows)

# Eine einfache Prüfung schützt vor unerwarteten Datenformen.
if data["group"].nunique() != 3:
    raise ValueError("Es sollten genau drei Gruppen vorhanden sein.")

# Diese Tabelle zeigt Mittelwert und Streuung kompakt.
summary = data.groupby("group")["satisfaction"].agg(["mean", "std", "count"])
summary = summary.round(1)
print("Kennzahlen je Gruppe:")
print(summary)

# Der Boxplot zeigt Median, Streuung und mögliche Ausreißer.
plt.figure(figsize=(8, 5))
ax = sns.boxplot(data=data, x="group", y="satisfaction", order=groups)

# Einzelpunkte machen Gruppengröße und Überlappung sichtbar.
sns.stripplot(
    data=data,
    x="group",
    y="satisfaction",
    order=groups,
    color="black",
    alpha=0.45,
    size=4,
    jitter=0.18,
    ax=ax,
)

ax.set_title("Zufriedenheit gezielt nach Filiale vergleichen")
ax.set_xlabel("Gruppe")
ax.set_ylabel("Zufriedenheit in Punkten")
plt.tight_layout()
plt.show()

```

### **3.1. Kategorien sicher vergleichen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_03_01.jpg?v=1784783828" width="250">



>* Gruppenunterschiede nicht vorschnell als Muster deuten
>* Gruppengröße und Repräsentativität zuerst prüfen

>* Fehlende Werte und Ausreißer verzerren Gruppenvergleiche.
>* Mehrere Kennzahlen statt nur Durchschnitt nutzen.

>* Störfaktoren können Gruppenunterschiede erklären
>* Muster vorsichtig beschreiben, Ursachen weiter prüfen



```
#@title Python-Code - Kategorien sicher vergleichen

# Wir vergleichen Kategorien mit Vorsicht.
# Gruppengrößen und Ausreißer verändern Mittelwerte.
# Die Grafik zeigt faire Vergleichsfragen.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Diese kleinen Daten enthalten absichtlich ungleiche Gruppen.
data = pd.DataFrame(
    {
        "store": ["A"] * 12 + ["B"] * 12 + ["C"] * 3,
        "satisfaction": [7, 8, 8, 7, 9, 8, 7, 8, 8, 9, 7, 8,
                         6, 6, 7, 6, 7, 6, 7, 6, 7, 6, 7, 6,
                         10, 10, 1],
    }
)

# Ein fehlender Wert macht Gruppe C noch unsicherer.
data.loc[26, "satisfaction"] = np.nan

# Mehrere Kennzahlen zeigen Risiken besser als ein Mittelwert.
summary = data.groupby("store").agg(
    count=("satisfaction", "count"),
    missing=("satisfaction", lambda values: values.isna().sum()),
    mean=("satisfaction", "mean"),
    median=("satisfaction", "median"),
)

# Gerundete Werte sind für den ersten Blick leichter lesbar.
summary = summary.round(2)
print("Gruppenübersicht: Anzahl, fehlende Werte, Mittelwert, Median")
print(summary.to_string())

# Die Warnung lenkt auf die wichtigste EDA-Frage.
small_groups = summary.index[summary["count"] < 5].tolist()
print("Vorsicht: Kleine Gruppen sind weniger belastbar:", small_groups)

# Boxplots zeigen Streuung und Ausreißer innerhalb jeder Kategorie.
fig, ax = plt.subplots(figsize=(7, 4))
data.boxplot(column="satisfaction", by="store", ax=ax)

# Titel und Achsen machen die Aussage der Grafik eindeutig.
ax.set_title("Kategorienvergleich: Verteilung statt nur Mittelwert")
ax.set_xlabel("Filiale")
ax.set_ylabel("Zufriedenheitspunkte")

# Der automatische Zusatztitel von pandas wird entfernt.
fig.suptitle("")
plt.show()

```

### **3.2. Korrelation kritisch prüfen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_03_02.jpg?v=1784783826" width="250">



>* Korrelation zeigt gemeinsame Veränderung, keine Ursache.
>* Prüfe Plausibilität, Drittvariablen und Alternativerklärungen.

>* Korrelationen brauchen Kontext und Alternativerklärungen
>* Datenherkunft, Verzerrungen und Untergruppen prüfen

>* Korrelationen mit Diagrammen und Gruppen prüfen
>* Unsicherheiten dokumentieren, keine Kausalität behaupten



```
#@title Python-Code - Korrelation kritisch prüfen

# Dieses Beispiel prüft Korrelationen kritisch.
# Eine Drittvariable kann Zusammenhänge erklären.
# Das Diagramm zeigt den scheinbaren Effekt.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Wir erzeugen kleine, nachvollziehbare Beispieldaten.
rng = np.random.default_rng(42)
temperature = rng.uniform(18, 34, 80)

# Beide Merkmale steigen mit der Temperatur.
ice_cream_sales = 20 + 4 * temperature + rng.normal(0, 8, 80)
swimming_accidents = 1 + 0.25 * temperature + rng.normal(0, 1.2, 80)

# Die Daten stehen in einer übersichtlichen Tabelle.
data = pd.DataFrame(
    {
        "temperature_c": temperature,
        "ice_cream_sales": ice_cream_sales,
        "swimming_accidents": swimming_accidents,
    }
)

# Eine einfache Prüfung schützt vor falschen Schlüssen.
if len(data) != 80:
    raise ValueError("Die Beispieldaten sollten 80 Zeilen enthalten.")

# Zuerst betrachten wir die auffällige direkte Korrelation.
direct_corr = data["ice_cream_sales"].corr(data["swimming_accidents"])

# Danach entfernen wir den linearen Temperatureinfluss.
sales_fit = np.polyfit(data["temperature_c"], data["ice_cream_sales"], 1)
accident_fit = np.polyfit(data["temperature_c"], data["swimming_accidents"], 1)

# Residuen zeigen, was nach Temperaturkontrolle übrig bleibt.
sales_residual = data["ice_cream_sales"] - np.polyval(
    sales_fit,
    data["temperature_c"],
)
accident_residual = data["swimming_accidents"] - np.polyval(
    accident_fit,
    data["temperature_c"],
)

# Diese Korrelation ist nach der Kontrolle deutlich vorsichtiger.
controlled_corr = sales_residual.corr(accident_residual)

print(f"Direkte Korrelation: {direct_corr:.2f}")
print(f"Nach Kontrolle der Temperatur: {controlled_corr:.2f}")
print("Merksatz: Korrelation ist ein Hinweis, kein Kausalbeweis.")

# Das Streudiagramm zeigt den scheinbaren Zusammenhang.
fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(
    data["ice_cream_sales"],
    data["swimming_accidents"],
    c=data["temperature_c"],
    cmap="coolwarm",
)

# Farbe macht die mögliche Drittvariable sichtbar.
ax.set_title("Korrelation kritisch prüfen: Temperatur als Drittvariable")
ax.set_xlabel("Eisverkäufe pro Tag")
ax.set_ylabel("Badeunfälle pro Tag")
fig.colorbar(scatter, ax=ax, label="Temperatur in °C")
plt.show()

```

### **3.3. Ergebnisse sicher dokumentieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_04/Lecture_A/image_03_03.jpg?v=1784783830" width="250">



>* Beobachtungen mit Kontext nachvollziehbar festhalten
>* Ausreißer und fehlende Werte vorsichtig bewerten

>* Zusammenhänge nicht vorschnell als Ursachen darstellen
>* Beschreibung, Deutung und Erklärung klar trennen

>* Entscheidungen und Unsicherheiten nachvollziehbar festhalten
>* Mögliche Verzerrungen transparent machen



```
#@title Python-Code - Ergebnisse sicher dokumentieren

# Dieses Beispiel dokumentiert EDA-Ergebnisse sicher.
# Es trennt Beobachtung, Risiko und vorsichtige Formulierung.
# Die Ausgabe zeigt eine kurze Analyse-Notiz.

import numpy as np
import pandas as pd

# Kleine Beispieldaten machen die Dokumentation nachvollziehbar.
data = pd.DataFrame(
    {
        "study_hours": [2, 3, 4, 5, 6, 7, 8, 9, 10, 30],
        "exam_score": [55, 58, 62, 65, 70, 74, 78, 82, 85, 88],
        "age_group": ["jung", "jung", "jung", "mittel", "mittel", "mittel", "mittel", "älter", "älter", "älter"],
        "income_eur": [1200, np.nan, np.nan, 2200, 2400, 2600, np.nan, 3100, 3300, 3500],
    }
)

# Einfache Prüfungen schützen vor stillen Datenproblemen.
row_count = len(data)
missing_income = int(data["income_eur"].isna().sum())
correlation = data["study_hours"].corr(data["exam_score"])

# Ein Ausreißer wird hier über den Interquartilsabstand markiert.
q1 = data["study_hours"].quantile(0.25)
q3 = data["study_hours"].quantile(0.75)
iqr = q3 - q1

upper_limit = q3 + 1.5 * iqr
outlier_count = int((data["study_hours"] > upper_limit).sum())
missing_by_group = data.groupby("age_group")["income_eur"].apply(lambda x: x.isna().mean())

# Die Notiz trennt Befund, Risiko und nächste Schritte.
print("Sichere EDA-Dokumentation")
print(f"Datenbasis: {row_count} Lernende, Variablen: Lernzeit, Note, Alter, Einkommen.")
print(f"Beobachtung: Korrelation Lernzeit-Note = {correlation:.2f}.")
print(f"Ausreißerprüfung: {outlier_count} Lernzeitwert über {upper_limit:.1f} Stunden.")
print(f"Fehlende Einkommen: {missing_income} von {row_count} Werten.")
print(f"Höchste Fehlquote nach Altersgruppe: {missing_by_group.idxmax()}.")
print("Vorsicht: Die Korrelation beschreibt keinen bewiesenen Ursache-Wirkungs-Effekt.")
print("Nächster Schritt: Ausreißer und fehlende Werte fachlich prüfen.")

```
