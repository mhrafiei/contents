## **1. Regeln und Muster**

### **1.1. Feste Regeln**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_A/image_01_01.jpg?v=1784781277" width="250">



>* Menschen legen klare Wenn-dann-Regeln fest
>* Entscheidungen sind nachvollziehbar, ohne Lernen aus Daten

>* Gut bei klaren, stabilen Bedingungen
>* ML unnötig bei bekannten Vorgaben

>* Komplexe Fälle erzeugen viele Ausnahmen
>* Regeln passen zu klaren Strukturen



```
#@title Python-Code - Feste Regeln

# Dieses Beispiel zeigt feste Regeln im Programm.
# Jede Entscheidung folgt einer vorher formulierten Bedingung.
# Die Ausgabe macht die Regelkette nachvollziehbar.

# Kleine Beispieldaten ersetzen hier echte Bankautomaten-Anfragen.
requests = [
    {"name": "Aylin", "card_valid": True, "pin_correct": True, "balance": 120},
    {"name": "Ben", "card_valid": False, "pin_correct": True, "balance": 200},
    {"name": "Clara", "card_valid": True, "pin_correct": False, "balance": 90},
    {"name": "David", "card_valid": True, "pin_correct": True, "balance": 30},
]

withdrawal_amount = 50

# Diese Funktion enthält nur feste, menschlich formulierte Regeln.
def decide_withdrawal(request, amount):
    if not request["card_valid"]:
        return "abgelehnt: Karte ungültig"

    if not request["pin_correct"]:
        return "abgelehnt: PIN falsch"

    if request["balance"] < amount:
        return "abgelehnt: Guthaben zu niedrig"

    return "ausgezahlt: alle Regeln erfüllt"

# Jede Zeile zeigt Eingabe, Regelprüfung und Ergebnis.
print("Feste Regeln beim Geldabheben:")

for request in requests:
    decision = decide_withdrawal(request, withdrawal_amount)
    print(f"{request['name']}: {decision}")

# Eine neue Regel ändert das Verhalten sofort und transparent.
daily_limit = 40

print("Neue Regel: Tageslimit 40 Euro.")

for request in requests[:1]:
    if withdrawal_amount > daily_limit:
        print(f"{request['name']}: abgelehnt: Tageslimit überschritten")
    else:
        print(f"{request['name']}: {decide_withdrawal(request, withdrawal_amount)}")

```

### **1.3. Wann Regeln genügen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_A/image_01_03.jpg?v=1784781278" width="250">



>* Klare Aufgaben brauchen oft nur Wenn-dann-Regeln
>* Regeln sind transparent und leicht änderbar

>* Regeln sichern nachvollziehbare, verlässliche Entscheidungen
>* Schutzgrenzen statt gelernter Wahrscheinlichkeiten nutzen

>* Einfache Aufgaben brauchen oft keine ML-Modelle
>* ML hilft bei komplexen, schwer formulierbaren Mustern



```
#@title Python-Code - Wann Regeln genügen

# Dieses Beispiel zeigt klare Wenn-dann-Regeln.
# Feste Grenzen ersetzen hier ein lernendes Modell.
# Die Ausgabe macht jede Entscheidung nachvollziehbar.

# Kleine Beispieldaten bleiben direkt im Programm sichtbar.
orders = [35, 80, 120, 250]
threshold = 100
discount_rate = 0.10

# Eine feste Regel entscheidet ohne Training aus Beispielen.
for order_value in orders:
    if order_value >= threshold:
        final_price = order_value * (1 - discount_rate)
        reason = "Rabatt, weil der Warenkorb mindestens 100 Euro erreicht."
    else:
        final_price = order_value
        reason = "Kein Rabatt, weil der Warenkorb unter 100 Euro liegt."

    print(f"{order_value} Euro -> {final_price:.2f} Euro: {reason}")

# Eine kurze Zusammenfassung verbindet Regel und Lernziel.
print("Fazit: Bei klaren Grenzen genügt eine transparente Regel.")

```

### **2.2. Daten und Modell**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_A/image_02_02.jpg?v=1784781286" width="250">



>* Daten sind Beispiele für Mustererkennung
>* Einseitige Daten begrenzen spätere ML-Leistung

>* Modelle lernen Muster aus Beispieldaten
>* Ihre Qualität hängt von passenden Daten ab

>* Daten helfen Modellen, Vorhersagen zu machen
>* Ergebnisse hängen von passenden Daten ab



```
#@title Python-Code - Daten und Modell

# Dieses Beispiel trennt Daten klar vom Modell.
# Ein Modell lernt Muster aus Beispielen.
# Neue Wohnungen zeigen die gelernte Schätzung.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import __version__ as sklearn_version

# Diese kleinen Daten beschreiben frühere Wohnungsangebote.
data = pd.DataFrame(
    {"area_sqm": [35, 45, 55, 65, 80, 95], "rent_eur": [520, 610, 760, 850, 1040, 1230]}
)

# Die Eingabe ist die Wohnfläche, die Ausgabe ist die Miete.
features = data[["area_sqm"]]
target = data["rent_eur"]

# Das Modell lernt eine einfache Beziehung aus den Daten.
model = LinearRegression()
model.fit(features, target)

# Neue Fälle sind keine Trainingsdaten, sondern Anwendungen des Modells.
new_flats = pd.DataFrame({"area_sqm": [50, 70, 90]})
predicted_rent = model.predict(new_flats)

# Diese Prüfung macht die erwartete Form der Vorhersagen sichtbar.
if len(predicted_rent) != len(new_flats):
    raise ValueError("Die Anzahl der Vorhersagen passt nicht zu den neuen Fällen.")

print(f"scikit-learn-Version: {sklearn_version}")
print(f"Daten: {len(data)} frühere Wohnungen mit Fläche und Miete.")
print(f"Modell: gelernte Steigung {model.coef_[0]:.1f} Euro pro Quadratmeter.")
print(f"Vorhersage: 70 qm kosten ungefähr {predicted_rent[1]:.0f} Euro.")

# Die Grafik zeigt Datenpunkte und die gelernte Modelllinie.
fig, ax = plt.subplots(figsize=(7, 4))
ax.scatter(data["area_sqm"], data["rent_eur"], label="Daten", color="tab:blue")

x_line = np.array([[30], [100]])
y_line = model.predict(x_line)
ax.plot(x_line.ravel(), y_line, label="Modell", color="tab:orange")

ax.scatter(new_flats["area_sqm"], predicted_rent, label="Neue Fälle", color="tab:green")
ax.set_title("Datenpunkte und daraus gelerntes Modell")
ax.set_xlabel("Wohnfläche in Quadratmetern")
ax.set_ylabel("Monatsmiete in Euro")

ax.legend()
plt.show()

```

## **3. ML Lebenszyklus**

### **3.1. Die richtige Frage**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_A/image_03_01.jpg?v=1784781289" width="250">



>* Erst Problem als Lernaufgabe prüfen
>* Klare Frage, Daten und Handlung festlegen

>* ML erkennt Muster in vielen Daten
>* Grenzen bei neuen oder sensiblen Entscheidungen

>* Erfolg und Fehlerfolgen vorher klar festlegen
>* Datenqualität schützt vor unrealistischen ML-Erwartungen



```
#@title Python-Code - Die richtige Frage

# Dieses Beispiel prüft eine mögliche ML-Frage.
# Es vergleicht Nutzen, Datenlage und Fehlerrisiko.
# Am Ende entsteht eine einfache Entscheidungshilfe.

import pandas as pd

# Drei Szenarien zeigen unterschiedliche Eignung für maschinelles Lernen.
scenarios = pd.DataFrame(
    {
        "scenario": ["Wartung", "Bewerbung", "Neue Krise"],
        "clear_question": [2, 1, 0],
        "data_quality": [2, 1, 0],
        "pattern_repetition": [2, 1, 0],
        "human_judgment_needed": [1, 2, 2],
        "error_impact": [1, 2, 2],
    }
)

# Höhere Werte bedeuten hier bessere Eignung für ein ML-Projekt.
scenarios["ml_fit_score"] = (
    scenarios["clear_question"]
    + scenarios["data_quality"]
    + scenarios["pattern_repetition"]
    - scenarios["human_judgment_needed"]
    - scenarios["error_impact"]
)

# Die Empfehlung übersetzt die Punktzahl in Alltagssprache.
recommendations = []
for score in scenarios["ml_fit_score"]:
    if score >= 3:
        recommendations.append("gute ML-Frage")
    elif score >= 1:
        recommendations.append("nur als Unterstützung")
    else:
        recommendations.append("erst klären")

scenarios["recommendation"] = recommendations

# Eine kleine Prüfung schützt vor unvollständigen Beispieldaten.
if len(scenarios) != 3:
    raise ValueError("Die Beispieldaten sollten genau drei Szenarien enthalten.")

# Die Ausgabe zeigt, dass nicht jede Automatisierung eine gute ML-Frage ist.
print("Bewertung einfacher ML-Fragen:")
for row in scenarios.itertuples(index=False):
    text = f"{row.scenario}: Score {row.ml_fit_score} -> {row.recommendation}"
    print(text)

print("Merksatz: ML passt besser bei klarer Frage, guten Daten und tragbaren Fehlern.")

```

### **3.3. Grenzen und Risiken**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_A/image_03_03.jpg?v=1784781291" width="250">



>* ML lernt nur aus vorhandenen Daten
>* Neue Situationen können zu Fehlern führen

>* ML-Ergebnisse können trügerisch objektiv wirken
>* Fehler, Benachteiligung und Einspruchsmöglichkeiten prüfen

>* Nutzen gegen mögliche Schäden abwägen
>* Datenschutz, Kontrolle und Verantwortung sichern



```
#@title Python-Code - Grenzen und Risiken

# Dieses Beispiel zeigt Grenzen eines ML-Modells.
# Trainingsdaten bestimmen, wo Vorhersagen zuverlässig wirken.
# Die Ausgabe vergleicht bekannte und neue Situationen.

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import sklearn

# Wir erzeugen einfache Sensordaten für Maschinenwartung.
rng = np.random.default_rng(42)
normal_temperature = rng.uniform(40, 80, 120)
normal_vibration = rng.uniform(1, 5, 120)

# Die Zielgröße beschreibt Tage bis zur Wartung.
normal_noise = rng.normal(0, 3, 120)
normal_days = 120 - normal_temperature - 8 * normal_vibration + normal_noise
normal_days = np.clip(normal_days, 1, 90)

# Diese Daten ähneln der Trainingsumgebung.
known_temperature = rng.uniform(40, 80, 40)
known_vibration = rng.uniform(1, 5, 40)
known_noise = rng.normal(0, 3, 40)
known_days = 120 - known_temperature - 8 * known_vibration + known_noise
known_days = np.clip(known_days, 1, 90)

# Diese Daten stammen aus einer veränderten Umgebung.
new_temperature = rng.uniform(75, 105, 40)
new_vibration = rng.uniform(4, 8, 40)
new_noise = rng.normal(0, 3, 40)
new_days = 150 - 1.2 * new_temperature - 10 * new_vibration + new_noise
new_days = np.clip(new_days, 1, 90)

# Wir prüfen, ob die Daten zusammenpassen.
if len(normal_temperature) != len(normal_days):
    raise ValueError("Trainingsdaten passen nicht zusammen.")

# Das Modell lernt nur aus der normalen Umgebung.
train_features = np.column_stack((normal_temperature, normal_vibration))
model = LinearRegression()
model.fit(train_features, normal_days)

# Wir bewerten bekannte und neue Situationen getrennt.
known_features = np.column_stack((known_temperature, known_vibration))
new_features = np.column_stack((new_temperature, new_vibration))
known_error = mean_absolute_error(known_days, model.predict(known_features))
new_error = mean_absolute_error(new_days, model.predict(new_features))

# Eine kleine Tabelle macht Nutzen und Grenze sichtbar.
result = pd.DataFrame(
    {
        "Szenario": ["ähnlich wie Training", "veränderte Umgebung"],
        "Fehler in Tagen": [round(known_error, 1), round(new_error, 1)],
    }
)

print(f"scikit-learn-Version: {sklearn.__version__}")
print("Ein Modell kann in bekannten Situationen nützlich sein.")
print(result.to_string(index=False))
print("Grenze: Bei veränderten Daten kann derselbe Ansatz deutlich schlechter werden.")

```
