## **1. Überwachtes Lernen**

### **1.1. Bekannte Zielwerte**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_01_01.jpg?v=1784782020" width="250">



>* Trainingsdaten enthalten Merkmale und bekannte Ergebnisse
>* Modelle lernen Muster für neue Fälle

>* Trainingsdaten enthalten richtige Antworten als Orientierung
>* Ohne Zielwerte fehlt die richtige Entscheidung

>* Bekannte Zielwerte zeigen überwachtes Lernen an
>* Kategorien oder Zahlen ermöglichen neue Vorhersagen



```
#@title Python-Code - Bekannte Zielwerte

# Dieses Beispiel zeigt bekannte Zielwerte im Datensatz.
# Zielwerte machen eine Aufgabe zum überwachten Lernen.
# Am Ende sehen wir Merkmale und passende Antworten.

import pandas as pd

# Jede Zeile beschreibt eine Wohnung mit bekanntem Mietpreis.
housing_data = pd.DataFrame(
    {
        "size_m2": [42, 55, 68, 80, 95],
        "rooms": [1, 2, 3, 3, 4],
        "rent_eur": [620, 760, 910, 1080, 1320],
    }
)

# Merkmale sind die beobachteten Eingaben für das Modell.
features = housing_data[["size_m2", "rooms"]]

# Der Zielwert ist die bekannte richtige Antwort.
target = housing_data["rent_eur"]

# Eine einfache Prüfung macht die Zuordnung sichtbar.
if len(features) != len(target):
    raise ValueError("Merkmale und Zielwerte passen nicht zusammen.")

# Die Ausgabe zeigt den Lösungsschlüssel im Training.
print("Überwachtes Lernen: Jede Zeile hat Merkmale und einen Zielwert.")
print(housing_data.to_string(index=False))
print("Zielwert: rent_eur, also ein bekannter Zahlenwert pro Wohnung.")
print("Ein Modell könnte später die Miete neuer Wohnungen vorhersagen.")

```

### **1.2. Klassifikation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_01_02.jpg?v=1784782021" width="250">



>* Modell ordnet Fälle bekannten Klassen zu
>* Trainingsdaten enthalten richtige Kategorien als Beispiele

>* Klassifikation ordnet Fälle bekannten Kategorien zu
>* Beispiele: Medizin, Betrug, Nachrichten, Bilder

>* Klassifikation sagt bekannte Kategorien voraus
>* Sinnvolle Klassen brauchen genügend Beispiele



```
#@title Python-Code - Klassifikation

# Dieses Beispiel zeigt Klassifikation mit bekannten Zielklassen.
# Ein Modell lernt aus beschrifteten Iris-Daten.
# Am Ende ordnet es neue Fälle Klassen zu.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Wir laden einen kleinen beschrifteten Datensatz.
iris = load_iris()
features = iris.data
target = iris.target

# Diese Prüfung macht die Grundannahme sichtbar.
if len(features) != len(target):
    raise ValueError("Merkmale und Zielklassen passen nicht zusammen.")

# Die Zielwerte sind vorgegebene Kategorien.
class_names = iris.target_names
class_count = len(class_names)

# Wir trennen Beispiele zum Lernen und Prüfen.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42, stratify=target
)

# Ein einfacher Entscheidungsbaum lernt die Zuordnung.
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Das Modell wählt für Testfälle bekannte Klassen aus.
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

# Drei neue Blüten werden ebenfalls klassifiziert.
new_flowers = X_test[:3]
new_predictions = model.predict(new_flowers)
new_labels = class_names[new_predictions]

print(f"scikit-learn-Version: {sklearn.__version__}")
print(f"Lernart: überwachtes Lernen mit {class_count} bekannten Klassen")
print(f"Beispielklassen: {', '.join(class_names)}")
print(f"Trefferquote auf Testdaten: {accuracy:.2f}")
print(f"Neue Fälle werden zugeordnet als: {', '.join(new_labels)}")

```

### **1.3. Werte vorhersagen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_01_03.jpg?v=1784782023" width="250">



>* Überwachtes Lernen kann Zahlenwerte vorhersagen
>* Bekannte Beispiele verbinden Merkmale mit Zielwerten

>* Messbare Ergebnisse aus Einflussgrößen schätzen
>* Bekannte Zielwerte ermöglichen richtige Vorhersagen

>* Bekannte Zielwerte kennzeichnen überwachtes Lernen
>* Vorhersagen sind Schätzungen, keine Sicherheiten



```
#@title Python-Code - Werte vorhersagen

# Dieses Beispiel zeigt überwachte Wertvorhersage.
# Bekannte Zielwerte ermöglichen eine Regression.
# Am Ende erscheint eine Mietpreis-Schätzung.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Kleine Beispieldaten beschreiben Wohnungen und bekannte Monatsmieten.
size_sqm = np.array([35, 45, 55, 65, 75, 85, 95, 110, 125, 140])
rent_eur = np.array([520, 610, 720, 810, 930, 1010, 1120, 1290, 1450, 1600])

# Scikit-learn erwartet Merkmale als zweidimensionale Tabelle.
features = size_sqm.reshape(-1, 1)
target = rent_eur

# Eine einfache Prüfung verhindert unpassende Tabellenformen.
if features.shape[0] != target.shape[0]:
    raise ValueError("Jede Wohnung braucht genau einen bekannten Zielwert.")

# Trainingsdaten enthalten Beispiele mit bereits bekannten richtigen Antworten.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.3, random_state=42
)

# Lineare Regression lernt den Zusammenhang zwischen Größe und Miete.
model = LinearRegression()
model.fit(X_train, y_train)

# Das Modell sagt Zahlenwerte für bisher ungesehene Wohnungen voraus.
predicted_test = model.predict(X_test)
new_flat = np.array([[100]])
predicted_rent = model.predict(new_flat)[0]

# Die mittlere Abweichung zeigt, dass Vorhersagen Schätzungen bleiben.
mae = mean_absolute_error(y_test, predicted_test)
print(f"scikit-learn-Version: {sklearn.__version__}")
print("Lernart: überwachtes Lernen, weil bekannte Mietpreise vorliegen.")
print("Aufgabentyp: Regression, weil ein Zahlenwert vorhergesagt wird.")
print(f"Mittlere Test-Abweichung: {mae:.0f} Euro.")
print(f"Geschätzte Miete für 100 m²: {predicted_rent:.0f} Euro.")

# Die Grafik verbindet bekannte Beispiele mit der gelernten Vorhersagelinie.
line_sizes = np.linspace(30, 145, 100).reshape(-1, 1)
line_rents = model.predict(line_sizes)

fig, ax = plt.subplots(figsize=(7, 4))
ax.scatter(size_sqm, rent_eur, label="bekannte Zielwerte")
ax.plot(line_sizes, line_rents, color="orange", label="gelernte Vorhersage")
ax.scatter(new_flat[0, 0], predicted_rent, color="red", label="neue Schätzung")

ax.set_title("Regression: Mietpreis als Wert vorhersagen")
ax.set_xlabel("Wohnungsgröße in m²")
ax.set_ylabel("Monatsmiete in Euro")
ax.legend()
plt.show()

```

## **2. Lernen ohne Zielwerte**

### **2.1. Gruppen in Daten**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_02_01.jpg?v=1784782025" width="250">



>* Clustering entdeckt Gruppen ohne vorgegebene Antworten
>* Menschen deuten gefundene Gruppen nachträglich

>* Clustering erkennt ähnliche Muster ohne feste Etiketten
>* Es ordnet unübersichtliche Daten sinnvoll

>* Cluster sind Vorschläge, keine sicheren Wahrheiten
>* Ergebnisse immer fachlich und kritisch prüfen



```
#@title Python-Code - Gruppen in Daten

# Dieses Beispiel zeigt Gruppen ohne Zielwerte.
# KMeans findet ähnliche Punkte durch Clustering.
# Die Grafik macht gefundene Gruppen sichtbar.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Wir erzeugen kleine Daten ohne Klassen-Spalte.
features, hidden_groups = make_blobs(
    n_samples=90, centers=3, cluster_std=0.8, random_state=42
)

# Diese Prüfung schützt vor unerwarteten Datenformen.
if features.shape != (90, 2):
    raise ValueError("Die Beispieldaten haben nicht die erwartete Form.")

# KMeans sucht drei Gruppen nur anhand der Merkmale.
model = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = model.fit_predict(features)

# Die echten Gruppen werden nicht zum Lernen verwendet.
found_count = len(set(cluster_labels))
print(f"scikit-learn-Version: {sklearn.__version__}")
print("Es gibt keine Zielwerte: KMeans sieht nur zwei Merkmale.")
print(f"Gefundene Gruppen: {found_count}")

# Die Farben zeigen die rechnerisch gefundenen Gruppen.
fig, ax = plt.subplots(figsize=(6, 4))
scatter = ax.scatter(
    features[:, 0], features[:, 1], c=cluster_labels, cmap="viridis", s=55
)

# Die roten Kreuze markieren die berechneten Gruppenmitten.
ax.scatter(
    model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
    c="red", marker="X", s=180, label="Gruppenmitten"
)

ax.set_title("Clustering: Gruppen in Daten ohne Zielwerte")
ax.set_xlabel("Merkmal 1")
ax.set_ylabel("Merkmal 2")
ax.legend()
plt.show()

```

### **2.3. Ungewöhnliche Muster erkennen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_02_03.jpg?v=1784782029" width="250">



>* Anomalien weichen deutlich vom Normalmuster ab
>* Sie zeigen Prüfbedarf, nicht automatisch Ursachen

>* Anomalien können Fehler oder wichtige Hinweise sein
>* Nützlich bei vielen normalen, wenigen Problemfällen

>* Anomalien brauchen menschliche Einordnung
>* Abweichungen statt Gruppen oder Zielwerte



```
#@title Python-Code - Ungewöhnliche Muster erkennen

# Wir erkennen ungewöhnliche Punkte ohne Zielwerte.
# IsolationForest lernt typische Muster aus Daten.
# Rote Punkte zeigen mögliche Anomalien.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import IsolationForest

# Diese Daten beschreiben normalen Stromverbrauch in Kilowattstunden.
rng = np.random.default_rng(42)
normal_day = rng.normal(loc=[18, 4], scale=[2.0, 0.7], size=(80, 2))

# Diese wenigen Nächte sind absichtlich ungewöhnlich hoch.
unusual_night = np.array([[2, 11], [3, 12], [1, 10]])
energy_data = np.vstack([normal_day, unusual_night])

# Eine einfache Prüfung schützt vor falschen Datenformen.
if energy_data.shape[1] != 2:
    raise ValueError("Die Daten brauchen genau zwei Merkmale.")

# Das Modell sucht Abweichungen vom typischen Muster.
model = IsolationForest(contamination=0.04, random_state=42)
model.fit(energy_data)

# Vorhersage minus eins bedeutet mögliche Anomalie.
predictions = model.predict(energy_data)
anomaly_mask = predictions == -1
normal_mask = predictions == 1

print(f"scikit-learn-Version: {sklearn.__version__}")
print(f"Beobachtungen insgesamt: {len(energy_data)}")
print(f"Als ungewöhnlich markiert: {int(anomaly_mask.sum())}")

# Die Grafik zeigt normale und ungewöhnliche Beobachtungen.
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(
    energy_data[normal_mask, 0],
    energy_data[normal_mask, 1],
    label="typisch",
    alpha=0.75,
)

ax.scatter(
    energy_data[anomaly_mask, 0],
    energy_data[anomaly_mask, 1],
    label="mögliche Anomalie",
    color="red",
    s=80,
)

ax.set_title("Anomalieerkennung beim Stromverbrauch")
ax.set_xlabel("Uhrzeit des Tages")
ax.set_ylabel("Verbrauch in kWh")
ax.legend()
plt.show()

```

### **3.3. Aufgaben passend zuordnen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_02/Lecture_B/image_03_03.jpg?v=1784782032" width="250">



>* Erst Ziel, Daten und Nutzen klären
>* Labels, Muster oder Belohnungen bestimmen Lernart

>* Erwartete Ausgabe bestimmt die Lernart.
>* Begründung erklärt Methode und passende Ergebnisform.

>* Ziel und Labels bestimmen die Lernart
>* Kontext und Fehlerkosten sorgfältig abwägen



```
#@title Python-Code - Aufgaben passend zuordnen

# Dieses Beispiel ordnet Alltagsaufgaben passenden Lernarten zu.
# Die Ergebnisform entscheidet über die sinnvolle Methode.
# Am Ende siehst du begründete Zuordnungen.

# Kleine Aufgabenbeschreibungen stehen direkt im Code.
tasks = [
    ("E-Mail als Spam oder Nicht-Spam erkennen", "feste Kategorie"),
    ("Lieferzeit eines Pakets in Minuten schätzen", "Zahlenwert"),
    ("Kundinnen nach ähnlichem Kaufverhalten gruppieren", "Gruppen finden"),
    ("Ungewöhnliche Login-Versuche entdecken", "seltene Abweichung"),
]

# Diese Regeln verbinden Ergebnisformen mit Lernarten.
learning_rules = {
    "feste Kategorie": "Klassifikation",
    "Zahlenwert": "Regression",
    "Gruppen finden": "Clustering",
    "seltene Abweichung": "Anomalieerkennung",
}

# Eine kurze Begründung macht die Zuordnung nachvollziehbar.
reasons = {
    "Klassifikation": "weil zwischen vorgegebenen Klassen entschieden wird",
    "Regression": "weil ein numerischer Wert vorhergesagt wird",
    "Clustering": "weil ähnliche Fälle ohne Labels gruppiert werden",
    "Anomalieerkennung": "weil seltene Auffälligkeiten gesucht werden",
}

# Die Ausgabe zeigt Aufgabe, Lernart und Begründung.
print("Aufgabe | passende Lernart | kurze Begründung")
for task, result_type in tasks:
    learning_type = learning_rules[result_type]
    reason = reasons[learning_type]
    print(f"{task} | {learning_type} | {reason}")

# Diese Merkhilfe fasst die wichtigste Entscheidungsfrage zusammen.
print("Merksatz: Frage zuerst nach der erwarteten Ausgabe.")

```
