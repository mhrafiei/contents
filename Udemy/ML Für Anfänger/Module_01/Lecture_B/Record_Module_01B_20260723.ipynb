## **1. NumPy Grundlagen**

### **1.1. Arrays erstellen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_01_01.jpg?v=1784779207" width="250">



>* NumPy-Arrays speichern gleichartige Werte effizient.
>* Form und Datentyp bewusst wählen.

>* Array-Form zeigt Bedeutung der Dimensionen
>* Regelmäßige Struktur verhindert falsche Berechnungen

>* Datentypen bestimmen Bedeutung, Genauigkeit und Speicherbedarf
>* Automatische Typwahl immer kritisch prüfen



```
#@title Python-Code - Arrays erstellen

# Dieses Beispiel erstellt kleine NumPy-Arrays.
# Form und Datentyp werden sichtbar geprüft.
# Die Ausgabe zeigt passende Array-Eigenschaften.

import numpy as np

# Eine Python-Liste wird in ein eindimensionales Array umgewandelt.
temperatures_celsius = np.array([18.5, 20.0, 19.5, 21.0], dtype=np.float64)

# Eine verschachtelte Liste erzeugt ein zweidimensionales Array.
sales_table = np.array([[12, 15, 11], [9, 14, 13]], dtype=np.int32)

# Wahrheitswerte können ebenfalls als eigenes Array gespeichert werden.
in_stock = np.array([True, False, True, True], dtype=bool)

# Diese Prüfung macht eine erwartete Tabellenform ausdrücklich sichtbar.
expected_shape = (2, 3)
if sales_table.shape != expected_shape:
    raise ValueError("Die Verkaufstabelle hat nicht die erwartete Form.")

print("Temperaturen:", temperatures_celsius.tolist())
print("Temperaturen: Form", temperatures_celsius.shape, "Datentyp", temperatures_celsius.dtype)
print("Verkaufstabelle: Form", sales_table.shape, "Datentyp", sales_table.dtype)
print("Verkaufstabelle als Liste:", sales_table.tolist())
print("Lagerstatus: Form", in_stock.shape, "Datentyp", in_stock.dtype)
print("Ergebnis: Werte, Form und Datentyp wurden bewusst festgelegt.")

```

### **1.2. Indexierung und Masken**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_01_02.jpg?v=1784779209" width="250">



>* Indexierung wählt gezielt Werte oder Ausschnitte
>* NumPy zählt Positionen ab null

>* Teilbereiche gezielt aus Arrays auswählen
>* Form des Ausschnitts für Berechnungen beachten

>* Masken filtern Daten nach Bedingungen
>* Sie zeigen Ausreißer und relevante Werte



```
#@title Python-Code - Indexierung und Masken

# Dieses Beispiel zeigt Indexierung und Masken.
# Wir wählen Werte nach Position und Bedingung.
# Die Ausgaben zeigen Form und gefilterte Werte.

import numpy as np

# Ein kleines Array stellt Messwerte in Grad Celsius dar.
temperatures = np.array([18.5, 21.0, 19.5, 24.0, 22.5, 17.0])

# Diese Prüfung macht die erwartete Array-Form sichtbar.
if temperatures.ndim != 1 or temperatures.size != 6:
    raise ValueError("Das Beispiel erwartet sechs Temperaturwerte.")

# Indexierung greift auf eine feste Position zu.
third_value = temperatures[2]

# Slicing wählt einen zusammenhängenden Ausschnitt aus.
workday_values = temperatures[1:5]

# Eine Maske markiert alle Werte über dem Grenzwert.
limit = 20.0
warm_mask = temperatures > limit

# Die Maske filtert nur die passenden Messwerte heraus.
warm_values = temperatures[warm_mask]

# Ein zweidimensionales Array zeigt Zeilen und Spalten.
sensor_table = temperatures.reshape(2, 3)

# Zeilen- und Spaltenindex wählen einen einzelnen Tabellenwert.
selected_cell = sensor_table[1, 2]

print("Temperaturen:", temperatures)
print("Form:", temperatures.shape, "Datentyp:", temperatures.dtype)
print("Dritter Wert:", third_value)
print("Ausschnitt Index 1 bis 4:", workday_values)
print("Maske über 20 Grad:", warm_mask)
print("Gefilterte Werte:", warm_values)
print("Tabelle 2x3:\n", sensor_table)
print("Wert in Zeile 1, Spalte 2:", selected_cell)

```

### **1.3. Rechnen mit Arrays**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_01_03.jpg?v=1784779211" width="250">



>* Arrays verarbeiten viele Werte gleichzeitig
>* Operationen laufen elementweise und effizient

>* Passende Array-Formen ermöglichen positionsweises Rechnen
>* Broadcasting erfordert sorgfältige Dimensionsprüfung

>* Kennzahlen verdichten viele Array-Werte.
>* Achsen bestimmen die fachliche Bedeutung.



```
#@title Python-Code - Rechnen mit Arrays

# Dieses Beispiel zeigt elementweises Rechnen mit NumPy-Arrays.
# Formen und Datentypen werden bewusst geprüft.
# Die Ausgabe vergleicht Rohwerte und berechnete Ergebnisse.

import numpy as np

# Drei Messwerte werden als kleines Array gespeichert.
temperatures_celsius = np.array([18.5, 20.0, 21.5], dtype=np.float64)

# Ein einzelner Korrekturwert wirkt auf jedes Element.
corrected_celsius = temperatures_celsius + 0.5

# Ein zweites Array gleicher Form wird positionsweise verrechnet.
reference_celsius = np.array([19.0, 19.5, 22.0], dtype=np.float64)
difference_celsius = corrected_celsius - reference_celsius

# Eine einfache Prüfung macht die passende Form sichtbar.
if corrected_celsius.shape != reference_celsius.shape:
    raise ValueError("Die Arrays müssen dieselbe Form haben.")

# Zusammenfassende Kennzahlen verdichten mehrere Werte.
average_difference = np.mean(difference_celsius)
largest_difference = np.max(difference_celsius)

print("Rohwerte in °C:", temperatures_celsius)
print("Form und Datentyp:", temperatures_celsius.shape, temperatures_celsius.dtype)
print("Nach Korrektur in °C:", corrected_celsius)
print("Abweichung zur Referenz in °C:", difference_celsius)
print("Mittlere Abweichung in °C:", round(average_difference, 2))
print("Größte Abweichung in °C:", round(largest_difference, 2))

```

## **2. pandas Tabellen**

### **2.1. Series und DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_02_01.jpg?v=1784779203" width="250">



>* Series speichern Werte mit einem Index
>* Beschriftungen ermöglichen robustes Ausrichten

>* DataFrames sind zweidimensionale Tabellen in pandas
>* Spalten können verschiedene Datentypen enthalten

>* Einzelne Spalten werden oft zu Series
>* Strukturwissen erleichtert Filtern und Analysieren



```
#@title Python-Code - Series und DataFrames

# Dieses Beispiel zeigt Series und DataFrames.
# Wir wählen, filtern und gruppieren Tabellendaten.
# Die Ausgabe vergleicht Spalte, Tabelle und Zusammenfassung.

import pandas as pd

# Eine Series ist eine beschriftete einzelne Spalte.
scores = pd.Series([82, 91, 76, 88], index=["Ana", "Ben", "Cem", "Dina"])

# Ein DataFrame enthält mehrere beschriftete Spalten.
students = pd.DataFrame(
    {
        "name": ["Ana", "Ben", "Cem", "Dina", "Eli"],
        "course": ["Python", "Python", "Daten", "Daten", "Python"],
        "score": [82, 91, 76, 88, 69],
        "hours": [5, 7, 4, 6, 3],
    }
)

# Diese Prüfung macht die erwartete Tabellenform sichtbar.
if students.shape != (5, 4):
    raise ValueError("Die Beispieltabelle sollte 5 Zeilen und 4 Spalten haben.")

# Eine einzelne Spalte aus dem DataFrame ist wieder eine Series.
score_column = students["score"]

# Mehrere Spalten bleiben zusammen ein DataFrame.
small_table = students[["name", "course", "score"]]

# Ein Wahrheitsfilter wählt nur passende Zeilen aus.
passed = small_table[small_table["score"] >= 80]

# Gruppieren fasst Zeilen nach Kurs zusammen.
summary = students.groupby("course", sort=True)["score"].mean().round(1)

print("Series mit Index:")
print(scores.head(3).to_string())
print("Ausgewählte DataFrame-Spalten:")
print(small_table.head(3).to_string(index=False))
print("Gefilterte Zeilen mit score >= 80:")
print(passed.to_string(index=False))
print("Durchschnittlicher score je course:")
print(summary.to_string())

```

### **2.2. Daten laden und filtern**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_02_02.jpg?v=1784779201" width="250">



>* Datenquellen als beschriftete Tabellen laden
>* Relevante Ausschnitte für Fragestellungen auswählen

>* Zeilen mit passenden Bedingungen auswählen
>* Filter immer an der Fragestellung prüfen

>* Relevante Spalten machen Tabellen übersichtlicher
>* Datentypen und Werte vor dem Filtern prüfen



```
#@title Python-Code - Daten laden und filtern

# Wir laden eine kleine Tabelle.
# Wir wählen Spalten und Zeilen.
# Das Ergebnis zeigt relevante Bestellungen.

import pandas as pd

# Die Daten entstehen direkt im Notebook.
orders = pd.DataFrame(
    {
        "customer": ["Ali", "Bea", "Cem", "Dana", "Eli", "Fia"],
        "region": ["Nord", "Süd", "Nord", "West", "Nord", "Süd"],
        "category": ["Buch", "Kurs", "Kurs", "Buch", "Kurs", "Buch"],
        "amount_eur": [18.5, 120.0, 75.0, 22.0, 140.0, 35.0],
        "paid": [True, True, False, True, True, False],
    }
)

# Eine einfache Prüfung macht die Tabelle verlässlicher.
expected_columns = {"customer", "region", "category", "amount_eur", "paid"}
if set(orders.columns) != expected_columns:
    raise ValueError("Die erwarteten Spalten fehlen.")

# Zuerst wählen wir nur fachlich wichtige Spalten.
selected_columns = ["customer", "region", "category", "amount_eur"]
compact_orders = orders[selected_columns]

# Danach formulieren wir Bedingungen direkt auf Spalten.
high_value = orders["amount_eur"] >= 50
north_region = orders["region"] == "Nord"
paid_orders = orders["paid"] == True

# Mehrere Bedingungen werden mit Klammern verbunden.
filtered_orders = compact_orders[high_value & north_region & paid_orders]

# Die Ausgabe bleibt klein und zeigt den Filtereffekt.
print("Ausgangstabelle: 6 Zeilen, 5 Spalten")
print("Filter: Region Nord, bezahlt, Betrag mindestens 50 Euro")
print(filtered_orders.to_string(index=False))

```

### **2.3. Gruppieren und Zählen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_02_03.jpg?v=1784779205" width="250">



>* Zeilen nach gemeinsamen Kategorien gruppieren
>* Gruppen zählen und übersichtlich zusammenfassen

>* Zeilen zählen oder vorhandene Werte zählen
>* Datenlücken und auffällige Gruppen erkennen

>* Erst filtern, dann nach Kategorien zählen
>* Gruppen prüfen und Muster sicher erkennen



```
#@title Python-Code - Gruppieren und Zählen

# Dieses Beispiel gruppiert eine kleine pandas Tabelle.
# Wir zählen Zeilen und vorhandene Werte.
# Das Ergebnis zeigt typische Datenlücken.

import pandas as pd

# Die Tabelle enthält Bestellungen mit einer fehlenden Bewertung.
orders = pd.DataFrame(
    {
        "region": ["Nord", "Süd", "Nord", "West", "Süd", "Nord"],
        "product": ["Buch", "Stift", "Buch", "Heft", "Buch", "Stift"],
        "rating": [5, 4, None, 3, 5, None],
    }
)

# Diese Prüfung macht die erwartete Tabellenform sichtbar.
if orders.shape != (6, 3):
    raise ValueError("Die Beispieltabelle sollte 6 Zeilen und 3 Spalten haben.")

# size zählt alle Zeilen pro Region.
rows_per_region = orders.groupby("region").size()

# count zählt nur vorhandene Werte in der Spalte rating.
ratings_per_region = orders.groupby("region")["rating"].count()

# Beide Zählweisen werden nebeneinander verglichen.
summary = pd.DataFrame(
    {
        "alle_zeilen": rows_per_region,
        "vorhandene_bewertungen": ratings_per_region,
    }
)

# Eine Differenz zeigt fehlende Bewertungen pro Gruppe.
summary["fehlende_bewertungen"] = (
    summary["alle_zeilen"] - summary["vorhandene_bewertungen"]
)

# Sortieren macht die Ausgabe stabil und gut lesbar.
summary = summary.sort_index()

print("Gruppieren nach Region: size zählt Zeilen, count zählt Werte.")
print(summary)

```

## **3. Tabellen prüfen**

### **3.1. Fehlende Werte erkennen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_03_01.jpg?v=1784779218" width="250">



>* Fehlende Werte früh erkennen
>* Ursachen klären, Verzerrungen vermeiden

>* Fehlende Werte nach Mustern untersuchen
>* Bedeutung im fachlichen Kontext bewerten

>* Fehlende Werte je nach Analyseziel behandeln
>* Vor Kennzahlen Vollständigkeit und Risiken prüfen



```
#@title Python-Code - Fehlende Werte erkennen

# Diese Übung zeigt fehlende Tabellenwerte.
# Pandas markiert Lücken mit True.
# Am Ende sehen wir klare Prüfzahlen.

import pandas as pd

# Eine kleine Tabelle enthält absichtlich einige Lücken.
data = {
    "student": ["Ada", "Ben", "Cem", "Dina", "Eli"],
    "points": [88, None, 74, 91, None],
    "group": ["A", "A", "B", None, "B"],
}

# Aus dem Wörterbuch entsteht ein übersichtlicher DataFrame.
students = pd.DataFrame(data)

# Diese Prüfung schützt vor unerwartet leeren Beispieldaten.
if students.shape != (5, 3):
    raise ValueError("Die Beispieltabelle hat nicht die erwartete Form.")

# isna erkennt fehlende Werte in jeder einzelnen Zelle.
missing_mask = students.isna()

# sum zählt True-Werte und liefert Lücken pro Spalte.
missing_per_column = missing_mask.sum()

# mean berechnet den Anteil fehlender Werte pro Spalte.
missing_share = (missing_mask.mean() * 100).round(1)

# Eine kompakte Übersicht verbindet Anzahl und Prozentanteil.
summary = pd.DataFrame(
    {"fehlend": missing_per_column, "anteil_prozent": missing_share}
)

print("Beispieltabelle mit möglichen Lücken:")
print(students.to_string(index=False))

print("Fehlende Werte pro Spalte:")
print(summary.to_string())

print("Wichtig: points.mean() ignoriert fehlende Werte automatisch.")
print("Mittelwert der vorhandenen Punkte:", round(students["points"].mean(), 1))

```

### **3.2. Datentypen prüfen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_03_02.jpg?v=1784779216" width="250">



>* Datentypen bestimmen die Interpretation von Spalten
>* Falsche Typen verfälschen spätere Kennzahlen

>* Datentypen müssen zur Bedeutung passen
>* Mischformen verursachen versteckte Analysefehler

>* Passende Datentypen machen Kennzahlen aussagekräftig
>* Prüfung zeigt nötige Bereinigungsschritte



```
#@title Python-Code - Datentypen prüfen

# Diese Übung prüft Datentypen in einer Tabelle.
# Falsche Typen verhindern sinnvolle Kennzahlen und Gruppierungen.
# Am Ende sehen wir passende Bereinigungsschritte.

import pandas as pd

# Die Tabelle enthält absichtlich typische kleine Datenprobleme.
sales = pd.DataFrame(
    {
        "order_date": ["01.02.2024", "05.02.2024", "12.03.2024", "18.03.2024"],
        "price_eur": ["19,99 €", "7,50 €", "unbekannt", "12,00 €"],
        "quantity": ["2", "1", "3", "n/a"],
        "paid": ["ja", "nein", "ja", "true"],
    }
)

# Zuerst betrachten wir die technischen Datentypen.
print("Datentypen vor der Bereinigung:")
print(sales.dtypes.to_string())

# Textwerte werden gezielt in passende Analyseformate umgewandelt.
cleaned = sales.copy()
cleaned["order_date"] = pd.to_datetime(cleaned["order_date"], format="%d.%m.%Y")
cleaned["price_eur"] = cleaned["price_eur"].str.replace(" €", "", regex=False)
cleaned["price_eur"] = cleaned["price_eur"].str.replace(",", ".", regex=False)
cleaned["price_eur"] = pd.to_numeric(cleaned["price_eur"], errors="coerce")

# Auch Mengen und Wahrheitswerte brauchen eindeutige Bedeutungen.
cleaned["quantity"] = pd.to_numeric(cleaned["quantity"], errors="coerce")
paid_map = {"ja": True, "nein": False, "true": True, "false": False}
cleaned["paid"] = cleaned["paid"].map(paid_map)

# Eine einfache Prüfung schützt vor stillen Analysefehlern.
expected_columns = ["order_date", "price_eur", "quantity", "paid"]
if list(cleaned.columns) != expected_columns:
    raise ValueError("Die erwarteten Spalten fehlen.")

# Jetzt zeigen die Datentypen die Rollen der Spalten besser.
print("Datentypen nach der Bereinigung:")
print(cleaned.dtypes.to_string())

# Fehlende Werte markieren Einträge, die noch geprüft werden sollten.
missing_counts = cleaned.isna().sum()
print("Fehlende Werte nach Umwandlung:")
print(missing_counts.to_string())

# Erst nach der Typprüfung sind Kennzahlen sinnvoll interpretierbar.
mean_price = cleaned["price_eur"].mean()
paid_share = cleaned["paid"].mean() * 100
print(f"Durchschnittspreis: {mean_price:.2f} Euro")
print(f"Bezahlte Bestellungen: {paid_share:.0f} Prozent")

```

### **3.3. Kurze Abschlussanalyse**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_01/Lecture_B/image_03_03.jpg?v=1784779220" width="250">



>* Wichtigste Befunde verständlich zusammenfassen
>* Nutzbarkeit, Einschränkungen und Muster benennen

>* Kennzahlen nur mit Datenqualität deuten
>* Unsicherheiten und mögliche Fehler klar benennen

>* Abschlussanalyse verbindet Prüfung und Entscheidung
>* Klare Einschätzung zeigt Nutzen und Grenzen



```
#@title Python-Code - Kurze Abschlussanalyse

# Diese Analyse verdichtet Kennzahlen und Datenprobleme.
# Eine kleine Tabelle zeigt typische Prüfentscheidungen.
# Am Ende entsteht eine kurze fachliche Einschätzung.

import pandas as pd

# Die Beispieldaten enthalten plausible und problematische Einträge.
sales = pd.DataFrame(
    {
        "order_id": [101, 102, 103, 104, 105, 106],
        "customer_id": ["C01", "C02", None, "C01", "C03", "C04"],
        "product_group": ["A", "B", "A", "B", "A", "B"],
        "revenue_eur": [120.0, 85.0, 999.0, 90.0, None, -15.0],
    }
)

# Diese Prüfungen verbinden Kennzahlen mit Datenqualität.
missing_customers = sales["customer_id"].isna().sum()
missing_revenue = sales["revenue_eur"].isna().sum()
negative_revenue = (sales["revenue_eur"] < 0).sum()

# Für Kennzahlen nutzen wir nur vorhandene Umsatzwerte.
valid_revenue = sales["revenue_eur"].dropna()
mean_revenue = valid_revenue.mean()
median_revenue = valid_revenue.median()
max_revenue = valid_revenue.max()

# Gruppierte Mittelwerte zeigen ein erstes fachliches Muster.
group_means = sales.groupby("product_group")["revenue_eur"].mean()
best_group = group_means.idxmax()
best_group_mean = group_means.max()

# Die Abschlussanalyse fasst Befunde knapp zusammen.
print("Kurze Abschlussanalyse der Verkaufstabelle")
print(f"Zeilen: {len(sales)}, Spalten: {sales.shape[1]}")
print(f"Fehlende Kundennummern: {missing_customers}")
print(f"Fehlende Umsätze: {missing_revenue}")
print(f"Negative Umsätze: {negative_revenue}")
print(f"Umsatz Mittelwert/Median: {mean_revenue:.1f} / {median_revenue:.1f} EUR")
print(f"Höchster Umsatz: {max_revenue:.1f} EUR")
print(f"Höchster Gruppenmittelwert: Gruppe {best_group} mit {best_group_mean:.1f} EUR")
print("Einschätzung: nutzbar, aber fehlende und auffällige Werte prüfen.")

```
