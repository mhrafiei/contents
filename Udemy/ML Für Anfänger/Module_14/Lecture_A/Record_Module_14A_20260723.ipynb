## **1. Neuronen und Schichten**

### **1.1. Das künstliche Neuron**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_01_01.jpg?v=1784806778" width="250">



>* Neuron gewichtet Eingaben und erzeugt Ausgaben
>* NumPy zeigt Gewichte, Bias und Rechenschritte

>* Gewichte steuern den Einfluss einzelner Eingaben
>* Bias und Training passen Entscheidungen an

>* Einzelne Neuronen erfassen vor allem lineare Muster
>* Schichten kombinieren Neuronen für komplexere Probleme



```
#@title Python-Code - Das künstliche Neuron

# Dieses Beispiel berechnet ein einzelnes künstliches Neuron.
# Eingaben, Gewichte und Bias werden sichtbar kombiniert.
# Die Ausgabe zeigt Zwischenwert und Aktivierung.

import numpy as np

# Kleine Eingaben stehen für drei einfache Merkmale.
inputs = np.array([0.8, 0.4, 0.1])
weights = np.array([0.6, -0.2, 0.9])
bias = 0.1

# Diese Prüfung verhindert unpassende Vektorlängen.
if inputs.shape != weights.shape:
    raise ValueError("Eingaben und Gewichte müssen gleich lang sein.")

# Das Neuron bildet zuerst eine gewichtete Summe.
weighted_inputs = inputs * weights
linear_value = np.sum(weighted_inputs) + bias

# Die Sigmoidfunktion macht daraus eine Ausgabe zwischen null und eins.
activation = 1 / (1 + np.exp(-linear_value))

# Die Ausgabe zeigt jeden Rechenschritt kompakt.
print("Eingaben:", np.round(inputs, 2))
print("Gewichte:", np.round(weights, 2))
print("Gewichtete Eingaben:", np.round(weighted_inputs, 2))
print("Bias:", round(bias, 2))
print("Zwischenwert z:", round(float(linear_value), 3))
print("Aktivierung sigmoid(z):", round(float(activation), 3))

```

### **1.2. Dense Schicht**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_01_02.jpg?v=1784806776" width="250">



>* Mehrere Neuronen verarbeiten dieselben Eingaben unterschiedlich
>* NumPy berechnet sie gemeinsam als Matrixoperation

>* Dense Schichten erzeugen neue Merkmale
>* Gewichte und Bias kombinieren Eingaben flexibel

>* Datenformen und Stapel sauber verfolgen
>* Vektorisierte Schichten transformieren Eingaben effizient



```
#@title Python-Code - Dense Schicht

# Diese Mini-Demo baut eine Dense Schicht.
# Matrixmultiplikation verbindet Eingaben mit mehreren Neuronen.
# Die Ausgabe zeigt Formen und berechnete Werte.

import numpy as np

# Drei Beispiele besitzen jeweils vier Eingabemerkmale.
inputs = np.array(
    [[70.0, 8.0, 20.0, 3.0], [45.0, 6.0, 35.0, 2.0], [90.0, 9.0, 10.0, 4.0]]
)

# Jedes der drei Neuronen hat vier eigene Gewichte.
weights = np.array(
    [[0.02, 0.50, -0.01], [0.30, -0.20, 0.40], [-0.04, 0.01, 0.03], [0.10, 0.20, -0.30]]
)

# Ein Bias verschiebt die Ausgabe jedes Neurons.
biases = np.array([0.5, -1.0, 0.2])

# Die Dense Schicht berechnet alle Neuronen gleichzeitig.
outputs = inputs @ weights + biases

# Diese Prüfung macht Formfehler früh verständlich.
if outputs.shape != (3, 3):
    raise ValueError("Die Ausgabe sollte drei Beispiele und drei Neuronen haben.")

# Runde Werte sind für den ersten Blick leichter lesbar.
rounded_outputs = np.round(outputs, 2)

print("Eingabeform: 3 Beispiele x 4 Merkmale")
print("Gewichtsform: 4 Merkmale x 3 Neuronen")
print("Ausgabeform: 3 Beispiele x 3 Neuronen")
print("Ausgaben der Dense Schicht:")
print(rounded_outputs)

```

### **1.3. Aktivierungsfunktionen verstehen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_01_03.jpg?v=1784806780" width="250">



>* Aktivierungen machen lineare Neuronen flexibel
>* Nichtlinearität ermöglicht Lernen komplexer Muster

>* Sigmoid, tanh und ReLU haben unterschiedliche Stärken
>* ReLU aktiviert nur passende Neuronen gezielt

>* Aktivierungen wirken elementweise auf ganze Arrays.
>* Ihre Wahl beeinflusst Lernfluss und Modellleistung.



```
#@title Python-Code - Aktivierungsfunktionen verstehen

# Wir vergleichen Aktivierungsfunktionen auf denselben Neuronensignalen.
# Nichtlinearität verändert Ausgaben einer Dense-Schicht elementweise.
# Die Grafik zeigt Sigmoid, Tanh und ReLU.

import numpy as np
import matplotlib.pyplot as plt

# Diese Werte stehen für Signale vor der Aktivierung.
pre_activation = np.linspace(-4.0, 4.0, 200)

# Jede Funktion wird elementweise auf das Array angewendet.
sigmoid = 1.0 / (1.0 + np.exp(-pre_activation))
tanh = np.tanh(pre_activation)
relu = np.maximum(0.0, pre_activation)

# Eine kleine Prüfung macht die Array-Formen nachvollziehbar.
if sigmoid.shape != pre_activation.shape:
    raise ValueError("Die Aktivierung muss die gleiche Form behalten.")

# Drei Beispielsignale zeigen konkrete Zahlen vor der Grafik.
sample_signals = np.array([-2.0, 0.0, 2.0])
sample_relu = np.maximum(0.0, sample_signals)
print("Beispielsignale vor Aktivierung:", sample_signals)
print("ReLU-Ausgaben danach:", sample_relu)

# Die Kurven zeigen unterschiedliche Reaktionsregeln eines Neurons.
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(pre_activation, sigmoid, label="Sigmoid")
ax.plot(pre_activation, tanh, label="Tanh")
ax.plot(pre_activation, relu, label="ReLU")

# Achsenbeschriftungen verbinden Mathematik mit Neuronenausgaben.
ax.set_title("Aktivierungsfunktionen als Reaktionsregeln")
ax.set_xlabel("Signal vor der Aktivierung")
ax.set_ylabel("Ausgabe nach der Aktivierung")
ax.legend()

plt.show()

```

## **2. Vorwärtsrechnung**

### **2.1. Stabile Softmax**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_02_01.jpg?v=1784806787" width="250">



>* Softmax wandelt Logits in Wahrscheinlichkeiten um
>* Hohe Werte zeigen relative Modellüberzeugung

>* Große Exponentialwerte können Softmax instabil machen
>* Maximalwert abziehen hält Wahrscheinlichkeiten unverändert

>* Stabile Softmax bewahrt relative Klassenunterschiede
>* Verlässliche Wahrscheinlichkeiten sichern Verlust und Gradienten



```
#@title Python-Code - Stabile Softmax

# Wir vergleichen naive und stabile Softmax.
# Große Logits zeigen das numerische Problem.
# Die stabile Variante liefert gültige Wahrscheinlichkeiten.

import numpy as np

# Diese Logits sind absichtlich sehr groß gewählt.
logits = np.array([[1000.0, 1001.0, 1002.0]])

# Die naive Rechnung kann durch exp überlaufen.
naive_exp = np.exp(logits)
naive_softmax = naive_exp / np.sum(naive_exp, axis=1, keepdims=True)

# Die stabile Rechnung verschiebt jede Zeile um ihr Maximum.
shifted_logits = logits - np.max(logits, axis=1, keepdims=True)
stable_exp = np.exp(shifted_logits)
stable_softmax = stable_exp / np.sum(stable_exp, axis=1, keepdims=True)

# Die Wahrscheinlichkeiten sollten zusammen eins ergeben.
probability_sum = np.sum(stable_softmax, axis=1)[0]

print("Logits:", logits[0].astype(int).tolist())
print("Naive Softmax:", np.round(naive_softmax[0], 4).tolist())
print("Verschobene Logits:", shifted_logits[0].astype(int).tolist())
print("Stabile Softmax:", np.round(stable_softmax[0], 4).tolist())
print("Summe der stabilen Wahrscheinlichkeiten:", round(float(probability_sum), 4))

```

### **2.2. Zwei Schichten**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_02_02.jpg?v=1784806792" width="250">



>* Erste Schicht erzeugt gelernte Zwischenmerkmale
>* Aktivierung ermöglicht komplexere Mustererkennung

>* Zweite Schicht erzeugt Klassen-Rohbewertungen
>* Vorherige Merkmale bestimmen ihre Qualität

>* Formen müssen zwischen Schichten passen
>* Softmax macht Rohwerte zu Klassenwahrscheinlichkeiten



```
#@title Python-Code - Zwei Schichten

# Dieses Beispiel zeigt Vorwärtsrechnung durch zwei Schichten.
# Dense-Schichten erzeugen Rohwerte aus Matrixmultiplikationen.
# Softmax macht daraus verständliche Klassenwahrscheinlichkeiten.

import numpy as np

# Kleine Eingaben stehen für drei Beispiele mit vier Merkmalen.
inputs = np.array([[1.0, 0.5, -1.0, 2.0], [0.0, 1.5, 1.0, -0.5], [2.0, -1.0, 0.5, 1.0]])

# Feste Gewichte machen die Rechnung vollständig nachvollziehbar.
weights_1 = np.array([[0.2, -0.4, 0.1], [0.7, 0.3, -0.2], [-0.5, 0.8, 0.6], [0.1, -0.1, 0.4]])
biases_1 = np.array([0.1, -0.2, 0.05])

# Die zweite Schicht erwartet genau drei Eingabewerte.
weights_2 = np.array([[0.3, -0.6], [-0.2, 0.5], [0.8, 0.1]])
biases_2 = np.array([0.0, 0.1])

# Diese Prüfung schützt vor vertauschten oder unpassenden Formen.
if inputs.shape[1] != weights_1.shape[0]:
    raise ValueError("Eingaben passen nicht zur ersten Schicht.")

# Die Ausgabe der ersten Schicht muss zur zweiten passen.
if weights_1.shape[1] != weights_2.shape[0]:
    raise ValueError("Die beiden Schichten passen nicht zusammen.")

# Erste Dense-Schicht plus ReLU erzeugt eine Zwischenrepräsentation.
hidden_raw = inputs @ weights_1 + biases_1
hidden_activated = np.maximum(0, hidden_raw)

# Zweite Dense-Schicht erzeugt Rohbewertungen für zwei Klassen.
logits = hidden_activated @ weights_2 + biases_2

# Softmax stabilisiert die Exponentialrechnung durch Zeilenverschiebung.
shifted_logits = logits - np.max(logits, axis=1, keepdims=True)
exp_scores = np.exp(shifted_logits)
probabilities = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

# Die Zeilensummen zeigen, dass Softmax Wahrscheinlichkeiten liefert.
print("Form Eingaben:", inputs.shape)
print("Form nach Schicht 1:", hidden_activated.shape)
print("Form nach Schicht 2:", logits.shape)
print("Softmax-Wahrscheinlichkeiten:")
print(np.round(probabilities, 3))
print("Zeilensummen:", np.round(probabilities.sum(axis=1), 3))

```

### **2.3. Verluste berechnen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_02_03.jpg?v=1784806789" width="250">



>* Verlust bewertet Softmax-Vorhersagen gegen Zielklassen
>* Kleiner Fehler liefert klares Trainingssignal

>* Kreuzentropie bestraft falsche Sicherheit stark
>* Durchschnittsverlust bewertet ganze Datenstapel stabil

>* Verlust bewertet Vorhersagen anhand bekannter Zielwerte
>* Messbarer Fehler steuert spätere Gewichtsänderungen



```
#@title Python-Code - Verluste berechnen

# Dieses Beispiel berechnet Verluste aus Softmax-Wahrscheinlichkeiten.
# Kreuzentropie bewertet die Wahrscheinlichkeit der richtigen Klasse.
# Am Ende vergleichen wir einzelne und mittlere Verluste.

import numpy as np

# Drei Beispiele haben bereits Softmax-Wahrscheinlichkeiten erhalten.
probabilities = np.array(
    [[0.70, 0.20, 0.10], [0.10, 0.80, 0.10], [0.05, 0.15, 0.80]]
)

# Die Zielwerte nennen pro Beispiel die richtige Klasse.
true_classes = np.array([0, 2, 1])

# Diese Prüfung verhindert unpassende Formen beim Indexieren.
if probabilities.shape[0] != true_classes.size:
    raise ValueError("Jedes Beispiel braucht genau einen Zielwert.")

# Wir wählen die vorhergesagte Wahrscheinlichkeit der richtigen Klasse aus.
sample_indices = np.arange(true_classes.size)
correct_probabilities = probabilities[sample_indices, true_classes]

# Ein kleiner Mindestwert schützt den Logarithmus vor null.
safe_probabilities = np.clip(correct_probabilities, 1e-12, 1.0)
losses = -np.log(safe_probabilities)

# Der mittlere Verlust fasst den Stapel zu einer Zahl zusammen.
average_loss = np.mean(losses)

print("Wahrscheinlichkeiten der richtigen Klassen:", np.round(correct_probabilities, 3))
print("Einzelverluste:", np.round(losses, 3))
print("Mittlerer Verlust:", round(float(average_loss), 3))
print("Hohe richtige Wahrscheinlichkeit bedeutet kleinen Verlust.")

```

## **3. Backpropagation mit NumPy**

### **3.1. Kettenregel verstehen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_03_01.jpg?v=1784806782" width="250">



>* Kettenregel zerlegt Fehler in lokale Einflüsse
>* Backpropagation verfolgt Gewichtsbeiträge rückwärts

>* Dense-Schichten verteilen Fehler rückwärts.
>* Lokale Ableitungen ergeben globale Gradienten.

>* Gespeicherte Zwischenschritte ermöglichen lokale Ableitungen
>* Rückwärts fließende Gradienten verbessern viele Parameter



```
#@title Python-Code - Kettenregel verstehen

# Dieses Beispiel zeigt die Kettenregel praktisch.
# Wir berechnen Gradienten einer Dense-Schicht.
# Numerische Prüfung bestätigt die Rückwärtsrechnung.

import numpy as np

# Kleine feste Werte machen jeden Schritt nachvollziehbar.
inputs = np.array([[1.0, -2.0, 0.5]])
weights = np.array([[0.2, -0.4], [0.7, 0.1], [-0.3, 0.5]])
biases = np.array([[0.1, -0.2]])
target = np.array([[0.0, 1.0]])

# Die Dense-Schicht berechnet gewichtete Summen plus Bias.
outputs = inputs @ weights + biases
loss = np.mean((outputs - target) ** 2)

# Dieser Gradient kommt vom Verlust zur Schichtausgabe zurück.
grad_outputs = 2.0 * (outputs - target) / outputs.size

# Die Kettenregel verteilt ihn auf Gewichte, Biases und Eingaben.
grad_weights = inputs.T @ grad_outputs
grad_biases = np.sum(grad_outputs, axis=0, keepdims=True)
grad_inputs = grad_outputs @ weights.T

# Eine numerische Ableitung prüft ein einzelnes Gewicht.
epsilon = 1e-5
weights_plus = weights.copy()
weights_minus = weights.copy()
weights_plus[0, 1] += epsilon
weights_minus[0, 1] -= epsilon

# Nur das geprüfte Gewicht wird minimal verändert.
loss_plus = np.mean((inputs @ weights_plus + biases - target) ** 2)
loss_minus = np.mean((inputs @ weights_minus + biases - target) ** 2)
numeric_gradient = (loss_plus - loss_minus) / (2.0 * epsilon)
analytic_gradient = grad_weights[0, 1]

# Die Formen zeigen, wohin jeder Gradient gehört.
print("Ausgabe der Schicht:", np.round(outputs, 3).tolist())
print("Verlust:", round(float(loss), 4))
print("Gradient Gewichte Form:", grad_weights.shape)
print("Gradient Biases Form:", grad_biases.shape)
print("Gradient Eingaben Form:", grad_inputs.shape)
print("Analytischer Gradient w[0,1]:", round(float(analytic_gradient), 6))
print("Numerischer Gradient w[0,1]:", round(float(numeric_gradient), 6))
print("Abweichung:", round(float(abs(analytic_gradient - numeric_gradient)), 10))

```

### **3.2. Gradienten numerisch prüfen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_03_02.jpg?v=1784806784" width="250">



>* Gradienten mit kleinen Gewichtsänderungen kontrollieren
>* Diagnosewerkzeug für plausible NumPy-Ableitungen

>* Parameter leicht ändern, Verlustunterschied berechnen
>* Mit Backpropagation vergleichen, Fehlerquellen erkennen

>* Nur gezielt und sparsam prüfen
>* Passende Schrittgröße schafft verlässliche Kontrolle



```
#@title Python-Code - Gradienten numerisch prüfen

# Wir prüfen einen Dense-Gradienten numerisch.
# Kleine Gewichtsänderungen schätzen die lokale Steigung.
# Beide Gradienten sollten fast gleich sein.

import numpy as np

# Diese Mini-Daten halten die Rechnung übersichtlich.
X = np.array([[0.2, -0.4], [0.7, 0.1], [-0.3, 0.5]])
y = np.array([[1.0], [0.0], [1.0]])

# Feste Parameter machen das Beispiel reproduzierbar.
weights = np.array([[0.3], [-0.2]])
bias = np.array([[0.1]])

# Die Formen müssen zur Dense-Schicht passen.
if X.shape[1] != weights.shape[0]:
    raise ValueError("Eingaben und Gewichte passen nicht zusammen.")

# Der mittlere quadratische Fehler ist leicht ableitbar.
def loss_for(current_weights, current_bias):
    predictions = X @ current_weights + current_bias
    errors = predictions - y
    return np.mean(errors ** 2)

# Backpropagation liefert den analytischen Gradienten.
predictions = X @ weights + bias
errors = predictions - y
batch_size = X.shape[0]

# Für MSE entsteht zuerst der Gradient nach den Ausgaben.
dloss_dpredictions = 2.0 * errors / batch_size
analytic_dweights = X.T @ dloss_dpredictions

# Wir prüfen genau ein Gewicht der Dense-Schicht.
row_index = 0
col_index = 0
epsilon = 1e-5

# Zentraler Differenzenquotient nutzt zwei Vorwärtsdurchläufe.
weights_plus = weights.copy()
weights_minus = weights.copy()
weights_plus[row_index, col_index] += epsilon
weights_minus[row_index, col_index] -= epsilon

# Aus den zwei Verlusten entsteht die numerische Steigung.
loss_plus = loss_for(weights_plus, bias)
loss_minus = loss_for(weights_minus, bias)
numeric_gradient = (loss_plus - loss_minus) / (2.0 * epsilon)

# Der Vergleich zeigt, ob die Implementierung plausibel ist.
analytic_gradient = analytic_dweights[row_index, col_index]
difference = abs(analytic_gradient - numeric_gradient)

print("Numerische Gradientenprüfung für eine Dense-Schicht")
print(f"Geprüftes Gewicht: weights[{row_index}, {col_index}]")
print(f"Analytischer Gradient: {analytic_gradient:.8f}")
print(f"Numerischer Gradient:  {numeric_gradient:.8f}")
print(f"Absolute Abweichung:   {difference:.10f}")
print(f"Prüfung bestanden:     {difference < 1e-7}")

```

### **3.3. Gewichte gezielt aktualisieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_14/Lecture_A/image_03_03.jpg?v=1784806786" width="250">



>* Gradienten zeigen den Beitrag jedes Gewichts
>* Backpropagation macht Fehler zu Lernschritten

>* Lernrate steuert sichere Gewichtsschritte
>* Balance hilft trotz Rauschen robust lernen

>* Gewichte wirken gemeinsam auf Vorhersagen
>* Geprüfte Gradienten machen Training nachvollziehbar



```
#@title Python-Code - Gewichte gezielt aktualisieren

# Wir aktualisieren Dense-Gewichte mit geprüften Gradienten.
# Die Lernrate steuert die Schrittgröße.
# Der Verlust soll nach dem Schritt sinken.

import numpy as np

# Kleine feste Daten machen jede Rechnung nachvollziehbar.
inputs = np.array([[1.0, -2.0], [0.5, 1.0], [-1.5, 2.0]])
targets = np.array([[1.0], [0.0], [1.0]])

# Startwerte sind bewusst klein und fest gewählt.
weights = np.array([[0.20], [-0.30]])
bias = np.array([[0.10]])

# Diese Prüfung verhindert unpassende Matrixformen.
if inputs.shape[1] != weights.shape[0]:
    raise ValueError("Eingaben und Gewichte passen nicht zusammen.")

# Vorwärtslauf und mittlerer quadratischer Fehler.
predictions = inputs @ weights + bias
errors = predictions - targets
loss_before = np.mean(errors ** 2)

# Gradienten zeigen die lokale Fehleränderung jedes Parameters.
sample_count = inputs.shape[0]
grad_predictions = 2.0 * errors / sample_count
grad_weights = inputs.T @ grad_predictions
grad_bias = np.sum(grad_predictions, axis=0, keepdims=True)

# Die Lernrate macht aus Gradienten kontrollierte Schritte.
learning_rate = 0.10
new_weights = weights - learning_rate * grad_weights
new_bias = bias - learning_rate * grad_bias

# Nach der Aktualisierung prüfen wir den neuen Verlust.
new_predictions = inputs @ new_weights + new_bias
new_errors = new_predictions - targets
loss_after = np.mean(new_errors ** 2)

# Kompakte Ausgabe zeigt Richtung, Schritt und Wirkung.
print("Gradient Gewicht 0:", round(float(grad_weights[0, 0]), 4))
print("Gradient Gewicht 1:", round(float(grad_weights[1, 0]), 4))
print("Alte Gewichte:", np.round(weights.ravel(), 4))
print("Neue Gewichte:", np.round(new_weights.ravel(), 4))
print("Verlust vorher:", round(float(loss_before), 4))
print("Verlust nachher:", round(float(loss_after), 4))
print("Gezielte Aktualisierung erfolgreich:", bool(loss_after < loss_before))

```
