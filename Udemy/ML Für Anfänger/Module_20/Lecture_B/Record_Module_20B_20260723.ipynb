```
#@title Python-Code - Problem auswählen

# Dieses Beispiel macht Problemauswahl praktisch greifbar.
# Wir bewerten Projektideen mit einfachen Kriterien.
# Am Ende entsteht eine begründete Priorisierung.

import pandas as pd

# Jede Zeile beschreibt eine mögliche Projektidee.
projects = pd.DataFrame(
    {
        "project": ["Support-Risiko", "Kundenzufriedenheit", "Betrugshinweis"],
        "clear_user_question": [5, 2, 4],
        "measurable_target": [5, 1, 4],
        "data_available": [4, 3, 3],
        "actionable_result": [4, 2, 5],
        "risk_manageable": [4, 5, 2],
    }
)

# Wir prüfen, ob die Tabelle die erwartete Form hat.
expected_columns = 6
if projects.shape[1] != expected_columns:
    raise ValueError("Die Projektbewertung hat unerwartete Spalten.")

# Diese Kriterien passen zur frühen ML-Projektplanung.
criteria = [
    "clear_user_question",
    "measurable_target",
    "data_available",
    "actionable_result",
    "risk_manageable",
]

# Der Durchschnitt zeigt eine einfache Gesamtbewertung.
projects["score"] = projects[criteria].mean(axis=1)
ranked_projects = projects.sort_values("score", ascending=False)

# Die beste Idee wird zusätzlich kurz begründet.
best_project = ranked_projects.iloc[0]
reason = "klare Frage, messbarer Zielwert und nutzbares Ergebnis"

print("Bewertungsskala: 1 = schwach, 5 = stark")
print(ranked_projects[["project", "score"]].round(2).to_string(index=False))
print("Beste Startidee:", best_project["project"])
print("Begründung:", reason)

```

### **1.2. Ziel klar definieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_01_02.jpg?v=1784822336" width="250">



>* Ziel beschreibt Entscheidung, Vorhersage und Nutzen

>* Zielwert eindeutig und messbar festlegen
>* Annahmen, Fälle und Zeitraum transparent machen

>* Erfolgskriterien passend zum Anwendungskontext wählen
>* Folgen, Fairness und Grenzen transparent berücksichtigen



```
#@title Python-Code - Ziel klar definieren

# Dieses Beispiel macht Projektziele messbar und prüfbar.
# Wir vergleichen unklare und klare Zieldefinitionen.
# Am Ende entsteht eine einfache Ziel-Checkliste.

import pandas as pd

# Jede Zeile beschreibt ein mögliches ML-Projektziel.
goals = pd.DataFrame(
    {
        "goal": ["Kundendaten analysieren", "Kündigung in 90 Tagen vorhersagen"],
        "user_question": [False, True],
        "target_value": [False, True],
        "time_window": [False, True],
        "success_metric": [False, True],
    }
)

# Diese Kriterien machen ein Ziel später umsetzbar.
criteria = ["user_question", "target_value", "time_window", "success_metric"]
goals["clarity_score"] = goals[criteria].mean(axis=1)

# Eine kurze Bewertung zeigt den Unterschied sofort.
labels = {False: "fehlt", True: "vorhanden"}
summary = goals.copy()
summary[criteria] = summary[criteria].replace(labels)

print("Zieldefinitionen im Vergleich:")
print(summary[["goal", "target_value", "time_window", "clarity_score"]])

# Das beste Ziel erfüllt alle vier Planungsfragen.
best_row = goals.loc[goals["clarity_score"].idxmax()]
print("Bestes Ziel:", best_row["goal"])
print("Erfolgskriterium: Ziel ist klar, wenn alle vier Kriterien erfüllt sind.")

```

### **1.3. Daten sauber beschreiben**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_01_03.jpg?v=1784822338" width="250">

```
#@title Python-Code - Daten sauber beschreiben

# Dieses Beispiel beschreibt Projektdaten vor der Modellierung.
# Wir prüfen Bedeutung, Herkunft und mögliche Datenleckage.
# Am Ende entsteht ein kompaktes Datensteckbrief-Ergebnis.

import pandas as pd

# Ein kleiner Beispieldatensatz steht für ein Kündigungsprojekt.
data = pd.DataFrame(
    {
        "contract_months": [3, 14, 28, 7, 41],
        "support_calls_last_30d": [2, 0, 1, 4, 0],
        "monthly_fee_eur": [29.9, 49.9, 39.9, 59.9, 19.9],
        "cancellation_reason": ["price", "none", "none", "service", "none"],
        "churned": [1, 0, 0, 1, 0],
    }
)

# Die Metadaten erklären jede Spalte fachlich.
metadata = pd.DataFrame(
    {
        "spalte": list(data.columns),
        "bedeutung": [
            "Vertragsdauer in Monaten",
            "Supportkontakte vor dem Stichtag",
            "Monatliche Gebühr in Euro",
            "Grund nach beobachteter Kündigung",
            "Zielwert: Kündigung beobachtet",
        ],
        "zeitpunkt": ["vorhersagezeitpunkt", "vorhersagezeitpunkt", "vorhersagezeitpunkt", "nachher", "nachher"],
    }
)

# Eine einfache Regel markiert mögliche Datenleckage.
feature_columns = [
    "contract_months",
    "support_calls_last_30d",
    "monthly_fee_eur",
    "cancellation_reason",
]

leakage_columns = metadata.loc[
    metadata["spalte"].isin(feature_columns) & (metadata["zeitpunkt"] == "nachher"),
    "spalte",
].tolist()

# Diese Prüfungen machen Annahmen sichtbar.
if len(data) != 5:
    raise ValueError("Der Beispieldatensatz sollte fünf Zeilen enthalten.")

if "churned" not in data.columns:
    raise ValueError("Der Zielwert churned fehlt im Datensatz.")

# Die Ausgabe zeigt einen kurzen Datensteckbrief.
print("Datensteckbrief für ein geplantes ML-Projekt")
print(f"Beobachtungseinheit: ein Kundenvertrag")
print(f"Zeilen und Spalten: {data.shape[0]} x {data.shape[1]}")
print(f"Zielwert: churned, beobachtet nach dem Vorhersagezeitpunkt")
print(f"Mögliche Datenleckage: {', '.join(leakage_columns)}")
print("Empfehlung: Leckage-Spalten vor dem Modelltraining entfernen.")

```

## **2. Workflow bauen**

### **2.1. Daten explorativ untersuchen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_02_01.jpg?v=1784822346" width="250">



>* Datenstruktur und Wertebereiche zuerst verstehen
>* Passung zur Fragestellung systematisch prüfen

>* Datenprobleme erkennen und Ursachen verstehen
>* Bereinigung und Modellierung begründet vorbereiten

>* Zusammenhänge und unausgewogene Gruppen erkennen
>* Datenlecks, Risiken und Annahmen dokumentieren



```
#@title Python-Code - Daten explorativ untersuchen

# Wir untersuchen Daten vor dem Modelltraining.
# Der Fokus liegt auf Qualität und Zielbezug.
# Die Ausgabe zeigt prüfbare EDA-Befunde.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Wir laden einen kleinen, eingebauten Klassifikationsdatensatz.
dataset = load_breast_cancer(as_frame=True)
features = dataset.frame.drop(columns="target")
target = dataset.frame["target"]

# Wir wählen wenige Spalten für eine übersichtliche Untersuchung.
selected_columns = ["mean radius", "mean texture", "mean area", "mean smoothness"]
data = features[selected_columns].copy()
data["target"] = target

# Eine einfache Validierung schützt vor falschen Annahmen.
row_count = len(data)
column_count = len(data.columns)
missing_count = int(data.isna().sum().sum())
duplicate_count = int(data.duplicated().sum())

# Klassenverteilung zeigt, ob das Ziel unausgewogen ist.
class_counts = data["target"].value_counts().sort_index()
class_share = (class_counts / row_count * 100).round(1)

# Korrelationen geben erste Hinweise, keine endgültigen Beweise.
correlations = data.corr(numeric_only=True)["target"].drop("target")
strongest_feature = correlations.abs().sort_values(ascending=False).index[0]
strongest_value = round(float(correlations[strongest_feature]), 2)

print(f"Datensatz: {row_count} Zeilen, {column_count} Spalten.")
print(f"Fehlende Werte: {missing_count}; doppelte Zeilen: {duplicate_count}.")
print(f"Klasse 0: {class_share.loc[0]}%; Klasse 1: {class_share.loc[1]}%.")
print(f"Stärkster Ziel-Zusammenhang: {strongest_feature}, r={strongest_value}.")

# Ein Boxplot macht Ausreißer und Gruppenunterschiede sichtbar.
fig, ax = plt.subplots(figsize=(7, 4))
data.boxplot(column="mean radius", by="target", ax=ax)

ax.set_title("Explorative Prüfung: Radius nach Zielklasse")
ax.set_xlabel("Zielklasse")
ax.set_ylabel("Mittlerer Radius")
fig.suptitle("")
plt.tight_layout()
plt.show()

```

### **2.2. Split und Baseline**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_02_02.jpg?v=1784822348" width="250">



>* Daten fachlich sinnvoll und getrennt aufteilen
>* Zeitbezug und Klassenverteilung korrekt beachten

>* Baseline zeigt Nutzen komplexerer Modelle.
>* Einfache Vergleiche verhindern falsche Erwartungen.

>* Split und Baseline früh dokumentieren
>* Vorverarbeitung nur auf Trainingsdaten lernen



```
#@title Python-Code - Split und Baseline

# Dieses Beispiel zeigt Split und Baseline.
# Wir vergleichen naive und gelernte Vorhersagen.
# Das Ergebnis macht faire Bewertung sichtbar.

import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Wir laden einen kleinen Klassifikationsdatensatz aus scikit-learn.
data = load_breast_cancer()
X = data.data
y = data.target

# Eine einfache Prüfung schützt vor stillen Datenproblemen.
if X.shape[0] != y.shape[0]:
    raise ValueError("Merkmale und Zielwerte haben unterschiedlich viele Zeilen.")

# Stratify hält die Klassenverteilung in beiden Teilmengen ähnlich.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Die Baseline sagt immer die häufigste Trainingsklasse voraus.
baseline = DummyClassifier(strategy="most_frequent")
baseline.fit(X_train, y_train)
baseline_predictions = baseline.predict(X_test)

# Skalierung wird nur auf Trainingsdaten gelernt, dank Pipeline.
model = make_pipeline(
    StandardScaler(), LogisticRegression(max_iter=1000, random_state=42)
)
model.fit(X_train, y_train)
model_predictions = model.predict(X_test)

# Beide Ansätze werden auf demselben Testset bewertet.
baseline_accuracy = accuracy_score(y_test, baseline_predictions)
model_accuracy = accuracy_score(y_test, model_predictions)
improvement = model_accuracy - baseline_accuracy

print(f"scikit-learn Version: {sklearn.__version__}")
print(f"Trainingszeilen: {len(y_train)}, Testzeilen: {len(y_test)}")
print(f"Baseline Accuracy: {baseline_accuracy:.3f}")
print(f"Logistische Regression Accuracy: {model_accuracy:.3f}")
print(f"Verbesserung gegenüber Baseline: {improvement:.3f}")

```

### **2.3. Modelle vergleichen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_02_03.jpg?v=1784822350" width="250">



>* Baseline als Maßstab für Mehrwert nutzen
>* Modelle fair mit gleichen Bedingungen vergleichen

>* Mehrere Metriken statt nur Genauigkeit nutzen
>* Fehler nach Gruppen und Mustern analysieren

>* Praxis zählt neben reiner Modellleistung.
>* Entscheidung begründen und Grenzen dokumentieren.



```
#@title Python-Code - Modelle vergleichen

# Wir vergleichen Modelle fair mit denselben Testdaten.
# Eine Baseline zeigt den einfachen Startpunkt.
# Die Tabelle macht die Entscheidung nachvollziehbar.

import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_breast_cancer

from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import recall_score

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier

# Wir laden einen kleinen Klassifikationsdatensatz aus scikit-learn.
data = load_breast_cancer()
X = data.data
y = data.target

# Eine kurze Prüfung verhindert stille Fehler im Workflow.
if X.shape[0] != y.shape[0]:
    raise ValueError("Merkmale und Zielwerte haben unterschiedlich viele Zeilen.")

# Alle Modelle bekommen denselben fairen Trainings-Test-Split.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Die Baseline sagt immer die häufigste Klasse voraus.
models = {
    "Baseline": DummyClassifier(strategy="most_frequent"),
    "Logistische Regression": make_pipeline(
        StandardScaler(), LogisticRegression(max_iter=1000, random_state=42)
    ),
    "Entscheidungsbaum": DecisionTreeClassifier(max_depth=4, random_state=42),
}

# Wir bewerten jedes Modell mit denselben drei Metriken.
rows = []
for model_name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    rows.append(
        {
            "Modell": model_name,
            "Accuracy": accuracy_score(y_test, predictions),
            "Balanced Accuracy": balanced_accuracy_score(y_test, predictions),
            "Recall wichtige Klasse": recall_score(y_test, predictions, pos_label=0),
        }
    )

# Die sortierte Tabelle zeigt Leistung und fachliche Perspektive.
results = pd.DataFrame(rows)
results = results.sort_values("Balanced Accuracy", ascending=False)
results = results.round(3)

print("scikit-learn Version:", sklearn.__version__)
print("Fairer Vergleich: gleicher Split, gleiche Daten, gleiche Metriken.")
print(results.to_string(index=False))

```

### **3.1. Fehler gezielt analysieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_03_01.jpg?v=1784822340" width="250">



>* Fehler nach Gruppen und Situationen untersuchen
>* Kennzahlen mit fachlicher Bewertung verbinden

>* Fehler nach Nutzungskontext und Folgen bewerten
>* Ungleich verteilte Fehler transparent dokumentieren

>* Analysewege und Grenzen nachvollziehbar dokumentieren
>* Fehlermuster, Risiken und Unsicherheiten klar benennen



```
#@title Python-Code - Fehler gezielt analysieren

# Wir analysieren Modellfehler nach Gruppen.
# Der Fokus liegt auf fairer Dokumentation.
# Das Ergebnis zeigt ungleich verteilte Fehler.

import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Wir erzeugen kleine, reproduzierbare Beispieldaten.
features, target = make_classification(
    n_samples=600,
    n_features=4,
    n_informative=3,
    n_redundant=0,
    random_state=42,
)

# Eine Gruppe ist absichtlich schwerer vorherzusagen.
rng = np.random.default_rng(42)
group = np.where(features[:, 0] > 0, "Gruppe A", "Gruppe B")
flip_mask = (group == "Gruppe B") & (rng.random(len(target)) < 0.25)

# Diese Störung simuliert ungleich gute Datenqualität.
noisy_target = target.copy()
noisy_target[flip_mask] = 1 - noisy_target[flip_mask]

# Wir prüfen eine zentrale Annahme vor dem Training.
if len(features) != len(noisy_target):
    raise ValueError("Merkmale und Zielwerte müssen gleich lang sein.")

# Die Gruppenspalte bleibt nur für die Fehleranalyse erhalten.
train_x, test_x, train_y, test_y, train_group, test_group = train_test_split(
    features,
    noisy_target,
    group,
    test_size=0.3,
    stratify=noisy_target,
    random_state=42,
)

# Ein einfaches Modell reicht für die Fehleranalyse.
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(train_x, train_y)

# Wir vergleichen Gesamtleistung und Gruppenfehler.
predicted = model.predict(test_x)
overall_accuracy = accuracy_score(test_y, predicted)
error = predicted != test_y

# Eine kleine Tabelle macht Muster sichtbar.
results = pd.DataFrame({"group": test_group, "error": error})
summary = results.groupby("group", sort=True)["error"].mean().reset_index()
summary["error_rate_percent"] = (summary["error"] * 100).round(1)

# Die Ausgabe bleibt kurz und berichtstauglich.
print(f"scikit-learn-Version: {sklearn.__version__}")
print(f"Gesamtgenauigkeit: {overall_accuracy:.2f}")
print("Fehlerrate nach Gruppe in Prozent:")
print(summary[["group", "error_rate_percent"]].to_string(index=False))
print("Dokumentation: Prüfe, ob höhere Fehler fachlich erklärbar sind.")

```

### **3.2. Verbesserungen gezielt umsetzen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_20/Lecture_B/image_03_02.jpg?v=1784822342" width="250">



>* Verbesserungen aus konkreten Fehlern ableiten
>* Maßnahmen begründen, testen und dokumentieren

>* Gesamtwerte zeigen nicht alle Risiken
>* Fairness und Datenschutz mitbewerten

>* Verbesserungsweg nachvollziehbar dokumentieren
>* Einfache, stabile Lösungen verantwortungsvoll wählen



```
#@title Python-Code - Verbesserungen gezielt umsetzen

# Dieses Beispiel zeigt gezielte Modellverbesserungen.
# Jede Änderung wird begründet und geprüft.
# Fairness und Reproduzierbarkeit bleiben sichtbar.

import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Wir erzeugen kleine, reproduzierbare Projektdaten.
features, target = make_classification(
    n_samples=600,
    n_features=4,
    n_informative=3,
    n_redundant=0,
    random_state=42,
)

# Eine Gruppe simuliert einen wichtigen Praxisvergleich.
group = np.where(features[:, 0] > 0, "Gruppe A", "Gruppe B")
if len(features) != len(target) or len(target) != len(group):
    raise ValueError("Daten, Zielwert und Gruppen müssen gleich lang sein.")

# Der Split bleibt stratifiziert und damit fairer vergleichbar.
X_train, X_test, y_train, y_test, group_train, group_test = train_test_split(
    features, target, group, test_size=0.3, stratify=target, random_state=42
)

# Skalierung wird nur auf Trainingsdaten gelernt.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Das Basismodell ist einfach und gut nachvollziehbar.
model = LogisticRegression(max_iter=300, random_state=42)
model.fit(X_train_scaled, y_train)
base_probabilities = model.predict_proba(X_test_scaled)[:, 1]

# Die Verbesserung passt nur die Entscheidungsgrenze an.
base_predictions = (base_probabilities >= 0.50).astype(int)
improved_predictions = (base_probabilities >= 0.45).astype(int)

# Wir vergleichen Gesamtwert und Gruppenwerte gemeinsam.
rows = []
for name, predictions in [("Basis", base_predictions), ("Verbessert", improved_predictions)]:
    overall = accuracy_score(y_test, predictions)
    score_a = accuracy_score(y_test[group_test == "Gruppe A"], predictions[group_test == "Gruppe A"])
    score_b = accuracy_score(y_test[group_test == "Gruppe B"], predictions[group_test == "Gruppe B"])
    rows.append([name, round(overall, 3), round(score_a, 3), round(score_b, 3)])

# Die Tabelle dokumentiert Nutzen und mögliche Nebenwirkungen.
result_table = pd.DataFrame(
    rows, columns=["Version", "Gesamt", "Gruppe A", "Gruppe B"]
)

print(f"scikit-learn-Version: {sklearn.__version__}")
print("Beobachtung: Gruppe B war im Test besonders wichtig zu prüfen.")
print("Änderung: Entscheidungsgrenze von 0.50 auf 0.45 gesenkt.")
print(result_table.to_string(index=False))
print("Dokumentation: Verbesserung nur übernehmen, wenn Nutzen und Fairness passen.")

```

```
#@title Python-Code - Projektbericht erstellen

# Dieser Bericht fasst ein kleines Modellprojekt zusammen.
# Die Vorlage verbindet Kennzahlen mit verantwortlicher Reflexion.
# Am Ende entsteht ein kompakter Projektbericht.

import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Wir laden einen kleinen, eingebauten Klassifikationsdatensatz.
data = load_breast_cancer(as_frame=True)
features = data.frame[data.feature_names]
target = data.frame["target"]

# Eine einfache Prüfung macht die Datenbasis nachvollziehbar.
if len(features) != len(target):
    raise ValueError("Merkmale und Zielwerte haben unterschiedliche Längen.")

# Die Trennung schützt die spätere Bewertung vor Datenleckage.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, stratify=target, random_state=42
)

# Skalierung wird nur mit Trainingsdaten gelernt.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ein einfaches Modell reicht für den Bericht als Beispiel.
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)

# Zwei Kennzahlen zeigen Nutzen und mögliche Fehlerfolgen.
predictions = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)
recall = recall_score(y_test, predictions)

# Eine kleine Tabelle sammelt Berichtspunkte statt Rohdaten.
report = pd.DataFrame(
    {
        "Berichtsteil": ["Daten", "Zielwert", "Erfolg", "Grenzen", "Reproduzierbarkeit"],
        "Kurznotiz": ["569 Fälle, 30 Merkmale", "Gutartig oder bösartig", f"Accuracy {accuracy:.2f}, Recall {recall:.2f}", "Nur Lehrdaten, keine Klinikfreigabe", "Split 75/25, random_state=42"],
    }
)

print(f"scikit-learn-Version: {sklearn.__version__}")
print("Projektbericht als kompakte Übergabe:")
print(report.to_string(index=False))
print("Fairness und Datenschutz: sensible Gruppen fehlen hier und müssten geprüft werden.")

```
