```
#@title Python-Code - Versionen prüfen

# Dieses Beispiel prüft wichtige Bibliotheksversionen.
# Versionen helfen bei reproduzierbarer Bildverarbeitung.
# Die Ausgabe zeigt die aktive Arbeitsumgebung.

import numpy as np
import cv2
from PIL import Image

# Wir lesen die Versionsnummern direkt aus den Bibliotheken.
pillow_version = Image.__version__
opencv_version = cv2.__version__
numpy_version = np.__version__

# Eine kleine synthetische RGB-Grafik prüft die Zusammenarbeit.
rgb_image = np.zeros((4, 4, 3), dtype=np.uint8)
rgb_image[:, :, 0] = 255

# Pillow kann das NumPy-Array als RGB-Bild interpretieren.
pillow_image = Image.fromarray(rgb_image, mode="RGB")

# OpenCV erwartet bei Farbumwandlungen oft BGR statt RGB.
bgr_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2BGR)

# Diese Prüfungen machen typische Annahmen sichtbar.
print(f"Pillow-Version: {pillow_version}")
print(f"OpenCV-Version: {opencv_version}")
print(f"NumPy-Version: {numpy_version}")
print(f"NumPy-Form: {rgb_image.shape}, Datentyp: {rgb_image.dtype}")
print(f"Pillow-Modus: {pillow_image.mode}, Größe: {pillow_image.size}")
print(f"RGB-Pixel links oben: {rgb_image[0, 0].tolist()}")
print(f"BGR-Pixel links oben: {bgr_image[0, 0].tolist()}")

```

### **1.2. Bilder anzeigen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_01_02.jpg?v=1784785974" width="250">



>* Bilder früh visuell auf Ladefehler prüfen

>* RGB und BGR korrekt unterscheiden
>* Achsen, Datentypen und Wertebereiche prüfen

>* Bilder stichprobenartig zur Qualitätssicherung prüfen
>* Anzeigeumgebungen beeinflussen die visuelle Kontrolle



```
#@title Python-Code - Bilder anzeigen

# Dieses Beispiel zeigt ein synthetisches Farbbild.
# Es vergleicht RGB und BGR beim Anzeigen.
# Die Ausgabe macht vertauschte Farbkanäle sichtbar.

import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image

# Wir erzeugen ein kleines RGB-Bild im Speicher.
height = 80
width = 120
rgb_image = np.zeros((height, width, 3), dtype=np.uint8)

# Drei Farbbereiche zeigen die Kanalreihenfolge deutlich.
rgb_image[:, :40] = [255, 0, 0]
rgb_image[:, 40:80] = [0, 255, 0]
rgb_image[:, 80:] = [0, 0, 255]

# Pillow kann direkt aus dem RGB-Array ein Bildobjekt bauen.
pillow_image = Image.fromarray(rgb_image, mode="RGB")
pillow_array = np.array(pillow_image)

# OpenCV nutzt häufig BGR und braucht eine Umwandlung.
bgr_image = cv2.cvtColor(pillow_array, cv2.COLOR_RGB2BGR)
converted_rgb = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# Eine einfache Prüfung schützt vor unerwarteten Formen.
if converted_rgb.shape != (height, width, 3):
    raise ValueError("Das Bild hat nicht die erwartete RGB-Form.")

# Kurze Zahlenwerte verbinden Anzeige und Array-Darstellung.
print("Form des Bildarrays:", converted_rgb.shape)
print("Datentyp der Pixel:", converted_rgb.dtype)
print("RGB-Pixel links:", converted_rgb[10, 10].tolist())
print("BGR-Pixel links:", bgr_image[10, 10].tolist())

# Matplotlib zeigt das korrekt zurückgewandelte RGB-Bild.
fig, ax = plt.subplots(figsize=(5, 3))
ax.imshow(converted_rgb)
ax.set_title("Synthetisches RGB-Bild nach OpenCV-Umwandlung")
ax.set_xlabel("Pixelspalte")
ax.set_ylabel("Pixelzeile")
plt.show()

```

### **1.3. Bildformate konvertieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_01_03.jpg?v=1784785972" width="250">



>* Formate passen zu unterschiedlichen Bildanforderungen
>* Konvertierung beeinflusst Qualität, Größe und Kompatibilität

>* Pillow, OpenCV und NumPy sinnvoll kombinieren
>* Farbraum, Kanalreihenfolge und Datentyp prüfen

>* Zielformat nach Zweck und Bildinhalt wählen
>* Einheitliche Formate sichern stabile Verarbeitung



```
#@title Python-Code - Bildformate konvertieren

# Dieses Beispiel konvertiert ein kleines synthetisches Bild.
# Pillow, OpenCV und NumPy arbeiten zusammen.
# Die Ausgabe zeigt Formate und Kanalreihenfolgen.

import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines RGB-Bild direkt im Speicher.
height = 64
width = 64
image_rgb = np.zeros((height, width, 3), dtype=np.uint8)

# Drei farbige Bereiche machen Kanalfehler gut sichtbar.
image_rgb[:, :22] = [220, 40, 40]
image_rgb[:, 22:43] = [40, 180, 70]
image_rgb[:, 43:] = [40, 80, 220]

# Pillow kann aus dem NumPy-Array ein Bildobjekt erstellen.
pillow_image = Image.fromarray(image_rgb, mode="RGB")
print(f"Pillow-Modus: {pillow_image.mode}, Größe: {pillow_image.size}")

# PNG bleibt verlustfrei, hier simuliert durch Bytes im Speicher.
png_bytes = pillow_image.tobytes()
print(f"RGB-Array: Form {image_rgb.shape}, Datentyp {image_rgb.dtype}")

# OpenCV erwartet für viele Funktionen die Reihenfolge BGR.
image_bgr = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR)
first_rgb_pixel = image_rgb[0, 0].tolist()
first_bgr_pixel = image_bgr[0, 0].tolist()

# Die Zahlen zeigen dieselbe Farbe in anderer Kanalreihenfolge.
print(f"Erster Pixel als RGB: {first_rgb_pixel}")
print(f"Erster Pixel als BGR: {first_bgr_pixel}")

# Für die Anzeige mit Matplotlib wechseln wir zurück zu RGB.
display_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
arrays_match = np.array_equal(image_rgb, display_rgb)
print(f"Rückkonvertierung korrekt: {arrays_match}")

# Die Figur zeigt das konvertierte Bild nach der Rückumwandlung.
fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(display_rgb)
ax.set_title("Synthetisches RGB-Bild nach BGR-Rückkonvertierung")

# Achsenwerte sind hier nicht wichtig für das Lernziel.
ax.set_xlabel("Pixelspalte")
ax.set_ylabel("Pixelzeile")
plt.show()

```

## **2. Pixel und Farben**

### **2.1. RGB und BGR**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_02_01.jpg?v=1784785964" width="250">



>* RGB-Pixel speichern Rot, Grün und Blau
>* Kanalwerte ermöglichen Farben und Bildanalyse

>* RGB und BGR unterscheiden Kanalreihenfolgen
>* Falsche Reihenfolge verfälscht sichtbare Farben

>* Kanalreihenfolge beeinflusst jede Pixelanalyse
>* Einheitliche Darstellung konsequent dokumentieren



```
#@title Python-Code - RGB und BGR

# Dieses Beispiel zeigt RGB und BGR.
# Kanalreihenfolgen verändern die sichtbaren Farben.
# Die Ausgabe vergleicht korrekte und falsche Interpretation.

import numpy as np
import matplotlib.pyplot as plt

# Wir bauen ein kleines synthetisches RGB-Bild.
height = 60
width = 180
rgb_image = np.zeros((height, width, 3), dtype=np.uint8)

# Drei Farbfelder füllen die RGB-Kanäle gezielt.
rgb_image[:, 0:60] = [255, 0, 0]
rgb_image[:, 60:120] = [0, 255, 0]
rgb_image[:, 120:180] = [0, 0, 255]

# OpenCV würde dieselben Zahlen als BGR interpretieren.
bgr_like_image = rgb_image.copy()
wrong_rgb_view = bgr_like_image
correct_rgb_view = bgr_like_image[:, :, ::-1]

# Wir prüfen die Form und die drei Kanalwerte.
if rgb_image.shape != (60, 180, 3):
    raise ValueError("Das Bild muss Höhe, Breite und drei Kanäle haben.")

print("Pixel links als RGB: R=255, G=0, B=0")
print("Gleiche Zahlen als BGR bedeuten: B=255, G=0, R=0")
print("Korrektur für Anzeige: Kanäle umdrehen mit image[:, :, ::-1]")

# Ein einziges Bild zeigt beide Interpretationen nebeneinander.
comparison = np.concatenate((wrong_rgb_view, correct_rgb_view), axis=1)
fig, ax = plt.subplots(figsize=(8, 2.5))

ax.imshow(comparison)
ax.set_title("Links: falsch als RGB angezeigt, rechts: BGR zu RGB korrigiert")
ax.set_xlabel("Pixelposition")
ax.set_ylabel("Pixelposition")

ax.text(90, 30, "falsch", color="white", ha="center", va="center")
ax.text(270, 30, "korrigiert", color="white", ha="center", va="center")
ax.set_xticks([])
ax.set_yticks([])

plt.show()

```

### **2.2. Graustufen und Alpha**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_02_02.jpg?v=1784785966" width="250">



>* Graustufen speichern Helligkeit statt Farbe
>* Weniger Daten, oft schnellere Verarbeitung

>* Graustufen gewichten Farben nach wahrgenommener Helligkeit
>* Wichtige Farbunterschiede können dabei verschwinden

>* Alpha-Kanäle speichern Transparenz pro Pixel
>* Kanalstruktur vor Analysen immer prüfen



```
#@title Python-Code - Graustufen und Alpha

# Dieses Beispiel zeigt Graustufen und Transparenz.
# Pixelkanäle werden sichtbar und vergleichbar.
# Das Diagramm zeigt Helligkeit mit Alpha.

import numpy as np
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines synthetisches RGBA-Bild.
height = 64
width = 64
x_values = np.linspace(0, 255, width, dtype=np.uint8)

# Rot, Grün und Blau bilden einfache Farbverläufe.
red_channel = np.tile(x_values, (height, 1))
green_channel = np.flipud(red_channel)
blue_channel = np.full((height, width), 80, dtype=np.uint8)

# Alpha beschreibt die Sichtbarkeit jedes Pixels.
alpha_channel = np.tile(x_values, (height, 1))
rgba_image = np.dstack((red_channel, green_channel, blue_channel, alpha_channel))

# Die Graustufe gewichtet Farben ähnlich der menschlichen Wahrnehmung.
gray_float = 0.299 * red_channel + 0.587 * green_channel + 0.114 * blue_channel
gray_image = np.rint(gray_float).astype(np.uint8)

# Wir prüfen die erwartete Kanalstruktur.
if rgba_image.shape != (height, width, 4):
    raise ValueError("Das RGBA-Bild hat nicht vier Kanäle.")

# Einige Pixel zeigen den Unterschied zwischen Helligkeit und Alpha.
sample_points = [(0, 0), (32, 32), (63, 63)]
print("Form des RGBA-Bildes:", rgba_image.shape)
print("Form des Graustufenbildes:", gray_image.shape)

# Die Ausgabe bleibt kurz und zeigt nur drei Beispielpixel.
for row, col in sample_points:
    rgba_values = rgba_image[row, col].tolist()
    gray_value = int(gray_image[row, col])
    print(f"Pixel ({row},{col}): RGBA={rgba_values}, Grau={gray_value}")

# Für die Anzeige kombinieren wir Graustufe und Alpha.
visible_gray = gray_image.astype(float) * (alpha_channel.astype(float) / 255.0)

# Ein einzelnes Bild zeigt transparente Bereiche dunkler.
fig, ax = plt.subplots(figsize=(5, 4))
image_plot = ax.imshow(visible_gray, cmap="gray", vmin=0, vmax=255)

# Achsen beschreiben Pixelpositionen im kleinen Bild.
ax.set_title("Synthetisches Graustufenbild mit Alpha")
ax.set_xlabel("x-Position in Pixeln")
ax.set_ylabel("y-Position in Pixeln")

# Eine Farbleiste erklärt die sichtbare Helligkeit.
fig.colorbar(image_plot, ax=ax, label="sichtbare Helligkeit")
plt.show()

```

### **2.3. Pixelwerte normalisieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_02_03.jpg?v=1784785968" width="250">



>* Pixelwerte auf gemeinsame Skala bringen
>* Algorithmen vergleichen Bilder dadurch stabiler

>* Normalisierung ändert Skala, nicht Farbbeziehungen
>* Methode beeinflusst Helligkeit und Kontrast

>* Normalisierung macht Histogramme und Operationen verständlicher
>* Rückumwandlung kann Bildwerte unerwartet verändern



```
#@title Python-Code - Pixelwerte normalisieren

# Dieses Beispiel normalisiert synthetische Pixelwerte.
# Die Skala wechselt von 0-255 zu 0-1.
# Histogramme zeigen dieselbe Verteilung auf neuer Skala.

import numpy as np
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines synthetisches Graustufenbild.
image_uint8 = np.array(
    [[0, 32, 64, 96], [128, 160, 192, 224], [255, 200, 120, 40]],
    dtype=np.uint8,
)

# Diese Prüfung schützt vor unerwarteten Pixelbereichen.
if image_uint8.min() < 0 or image_uint8.max() > 255:
    raise ValueError("Die Pixelwerte müssen zwischen 0 und 255 liegen.")

# Normalisierung macht aus ganzen Zahlen vergleichbare Gleitkommazahlen.
image_normalized = image_uint8.astype(np.float32) / 255.0

# Wir vergleichen wenige Werte vor und nach der Normalisierung.
print("Originalwerte:", image_uint8[0].tolist())
print("Normalisiert:", np.round(image_normalized[0], 3).tolist())
print("Originalbereich:", int(image_uint8.min()), "bis", int(image_uint8.max()))
print("Normalisierter Bereich:", round(float(image_normalized.min()), 3), "bis", round(float(image_normalized.max()), 3))

# Das Histogramm zeigt die Verteilung auf der normalisierten Skala.
fig, ax = plt.subplots(figsize=(6, 4))
ax.hist(image_normalized.ravel(), bins=8, range=(0, 1), color="steelblue")

ax.set_title("Histogramm normalisierter Pixelwerte")
ax.set_xlabel("Normalisierte Intensität")
ax.set_ylabel("Anzahl Pixel")
plt.show()

```

## **3. Bildoperationen automatisieren**

### **3.1. Zuschneiden und Skalieren**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_03_01.jpg?v=1784785962" width="250">



>* Zuschnitt fokussiert das wichtige Bildmotiv
>* Skalierung macht Bilder einheitlich verarbeitbar

>* Zuschnittregel passend zum Motiv wählen
>* Ergebnisse fachlich mit Beispielen prüfen

>* Seitenverhältnis bewahren, Verzerrungen vermeiden
>* Klare Skalierregeln schaffen konsistente Bilddaten



```
#@title Python-Code - Zuschneiden und Skalieren

# Dieses Beispiel zeigt automatisches Zuschneiden und Skalieren.
# Ein synthetisches Bild macht die Schritte sichtbar.
# Am Ende haben alle Bilder gleiche Zielmaße.

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Wir erzeugen ein kleines synthetisches RGB-Bild.
height = 80
width = 120
image = np.full((height, width, 3), 235, dtype=np.uint8)

# Ein farbiges Motiv liegt bewusst in der Bildmitte.
image[20:60, 35:85, 0] = 40
image[20:60, 35:85, 1] = 120
image[20:60, 35:85, 2] = 220

# Der zentrale Zuschnitt entfernt gleichmäßig Randbereiche.
crop_size = 64
start_y = (height - crop_size) // 2
start_x = (width - crop_size) // 2

# NumPy schneidet mit Zeilen und Spalten zu.
cropped = image[start_y:start_y + crop_size, start_x:start_x + crop_size]
if cropped.shape != (64, 64, 3):
    raise ValueError("Der Zuschnitt hat nicht die erwartete Form.")

# Pillow skaliert den Zuschnitt auf eine feste Zielgröße.
target_size = (32, 32)
pil_cropped = Image.fromarray(cropped)
resized = pil_cropped.resize(target_size, Image.Resampling.BILINEAR)

# Zur Anzeige vergrößern wir nur die fertige Vorschau.
preview = resized.resize((128, 128), Image.Resampling.NEAREST)
preview_array = np.array(preview)

print(f"Original: {image.shape[1]} x {image.shape[0]} Pixel")
print(f"Zuschnitt: {cropped.shape[1]} x {cropped.shape[0]} Pixel")
print(f"Skaliert: {resized.size[0]} x {resized.size[1]} Pixel")

# Die einzelne Grafik zeigt das einheitlich vorbereitete Ergebnis.
fig, ax = plt.subplots(figsize=(4, 4))
ax.imshow(preview_array)
ax.set_title("Synthetisches Bild nach Zuschnitt und Skalierung")
ax.set_xlabel("Pixelspalte in der Vorschau")
ax.set_ylabel("Pixelzeile in der Vorschau")
ax.set_xticks([])
ax.set_yticks([])
plt.show()

```

### **3.2. Filter und Kanten**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_03_02.jpg?v=1784785958" width="250">



>* Filter machen Bildvorverarbeitung vergleichbar und automatisierbar
>* Filterstärke beeinflusst Details, Kanten und Rauschen

>* Kanten zeigen wichtige Bildübergänge und Strukturen
>* Glättung reduziert falsche Kanten durch Störungen

>* Einheitliche Pipeline für vergleichbare Bildverarbeitung
>* Schritte dokumentieren und Ergebnisse visuell prüfen



```
#@title Python-Code - Filter und Kanten

# Dieses Beispiel automatisiert Filter und Kantenerkennung.
# Eine synthetische Aufnahme macht die Schritte kontrollierbar.
# Die Ausgabe zeigt eine einheitliche Vorverarbeitung.

import numpy as np
import cv2
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines synthetisches Graustufenbild.
image = np.full((96, 96), 35, dtype=np.uint8)
cv2.rectangle(image, (18, 22), (76, 70), 180, thickness=-1)
cv2.circle(image, (48, 46), 16, 235, thickness=-1)

# Deterministisches Rauschen simuliert eine unruhige Aufnahme.
rng = np.random.default_rng(42)
noise = rng.normal(0, 18, image.shape)
noisy_image = np.clip(image.astype(np.float32) + noise, 0, 255)
noisy_image = noisy_image.astype(np.uint8)

# Dieselbe Pipeline wird als feste Funktion formuliert.
def preprocess_image(input_image):
    blurred = cv2.GaussianBlur(input_image, (5, 5), 0)
    edges = cv2.Canny(blurred, threshold1=60, threshold2=140)
    return blurred, edges

# Eine einfache Prüfung schützt vor falschen Bildformen.
if noisy_image.ndim != 2:
    raise ValueError("Dieses Beispiel erwartet ein Graustufenbild.")

# Die Pipeline liefert geglättete Pixel und Kantenpixel.
blurred_image, edge_image = preprocess_image(noisy_image)
edge_share = 100 * np.count_nonzero(edge_image) / edge_image.size

print("Synthetisches Bild: 96 x 96 Pixel.")
print(f"Kantenanteil nach Glättung: {edge_share:.2f} Prozent.")
print("Pipeline: Rauschen -> Gauß-Filter -> Canny-Kanten.")

# Für eine Achse kombinieren wir drei Ansichten nebeneinander.
combined_image = np.hstack((noisy_image, blurred_image, edge_image))

fig, ax = plt.subplots(figsize=(9, 3))
ax.imshow(combined_image, cmap="gray", vmin=0, vmax=255)
ax.set_title("Synthetisch: verrauscht, geglättet, Kanten")
ax.set_xlabel("Drei Pipeline-Schritte nebeneinander")
ax.set_ylabel("Pixelzeile")
ax.set_xticks([48, 144, 240])
ax.set_xticklabels(["verrauscht", "geglättet", "Kanten"])
plt.show()

```

### **3.3. Bildordner vorbereiten**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_A/image_03_03.jpg?v=1784785960" width="250">



>* Unterschiedliche Bildquellen einheitlich vorbereiten
>* Prüfen, skalieren und geordnet speichern

>* Ziele der Bildvorbereitung vorher festlegen
>* Einheitliche, reproduzierbare Pipeline verwenden

>* Dateien klar benennen und sinnvoll ordnen
>* Problemfälle prüfen und Qualität kontrollieren



```
#@title Python-Code - Bildordner vorbereiten

# Wir bereiten einen synthetischen Bildordner vor.
# Alle Bilder erhalten gleiche Größe und Kanäle.
# Die Ausgabe zeigt die vereinheitlichte Bildsammlung.

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Diese Liste simuliert Dateien aus einem kleinen Bildordner.
rng = np.random.default_rng(42)
raw_images = []

# Ein RGB-Bild mit anderer Größe wird erzeugt.
raw_images.append(rng.integers(0, 256, size=(40, 60, 3), dtype=np.uint8))

# Ein Graustufenbild simuliert fehlende Farbkanäle.
raw_images.append(rng.integers(0, 256, size=(55, 35), dtype=np.uint8))

# Ein RGBA-Bild simuliert einen transparenten Hintergrund.
raw_images.append(rng.integers(0, 256, size=(45, 45, 4), dtype=np.uint8))

# Die Zielgröße macht alle Bilder vergleichbar.
target_size = (32, 32)
processed_images = []

# Jedes Bild wird nach RGB konvertiert und skaliert.
for image_array in raw_images:
    pil_image = Image.fromarray(image_array)
    rgb_image = pil_image.convert("RGB")
    resized_image = rgb_image.resize(target_size, Image.Resampling.BILINEAR)
    processed_images.append(np.array(resized_image, dtype=np.uint8))

# Ein Stapel ist praktisch für spätere Analysen.
image_batch = np.stack(processed_images, axis=0)

# Diese Prüfung verhindert versteckte Formfehler.
expected_shape = (len(raw_images), 32, 32, 3)
if image_batch.shape != expected_shape:
    raise ValueError("Die vorbereiteten Bilder haben eine unerwartete Form.")

print("Vorbereitung abgeschlossen: 3 synthetische Bilder.")
print(f"Einheitliche Stapelform: {image_batch.shape}")
print(f"Datentyp und Wertebereich: {image_batch.dtype}, 0 bis 255")

# Die drei Bilder werden nebeneinander in einer Achse gezeigt.
preview = np.concatenate(processed_images, axis=1)
plt.figure(figsize=(6, 2))
plt.imshow(preview)
plt.title("Vorverarbeitete synthetische Bilder: RGB, 32 x 32 Pixel")
plt.xlabel("Bildindex nebeneinander")
plt.ylabel("Pixelzeile")
plt.xticks([16, 48, 80], ["Bild 1", "Bild 2", "Bild 3"])
plt.yticks([0, 16, 31])
plt.show()

```
