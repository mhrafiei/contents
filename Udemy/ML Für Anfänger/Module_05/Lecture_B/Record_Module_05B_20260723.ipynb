## **1. Zeitreihen erzeugen**

### **1.1. Zeitachse und Abtastung**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_01_01.jpg?v=1784787029" width="250">



>* Zeitreihen ordnen Messwerte entlang der Zeit.
>* Zeitpunkte machen Veränderungen und Störungen sichtbar.

>* Abtastung macht kontinuierliche Signale digital messbar
>* Passende Abstände verhindern verlorene Signaldetails

>* Zeitdauer und Abtastung bewusst wählen
>* Genauigkeit, Übersichtlichkeit und Datenmenge ausbalancieren



```
#@title Python-Code - Zeitachse und Abtastung

# Wir erzeugen eine einfache Zeitachse.
# Die Abtastrate bestimmt die Messpunkte.
# Die Grafik zeigt grobe und feine Abtastung.

import numpy as np
import matplotlib.pyplot as plt

# Diese Werte legen Dauer und Abtastraten fest.
duration_seconds = 2.0
low_rate_hz = 8
high_rate_hz = 80

# Die Zeitachsen enthalten gleichmäßig verteilte Messzeitpunkte.
low_time = np.arange(0, duration_seconds, 1 / low_rate_hz)
high_time = np.arange(0, duration_seconds, 1 / high_rate_hz)

# Eine einfache Schwingung dient als bekanntes Beispielsignal.
frequency_hz = 2
low_signal = np.sin(2 * np.pi * frequency_hz * low_time)
high_signal = np.sin(2 * np.pi * frequency_hz * high_time)

# Eine kurze Prüfung macht die Anzahl der Messpunkte sichtbar.
print(f"Grobe Abtastung: {len(low_time)} Messpunkte bei {low_rate_hz} Hz")
print(f"Feine Abtastung: {len(high_time)} Messpunkte bei {high_rate_hz} Hz")
print(f"Beobachtete Dauer: {duration_seconds:.1f} Sekunden")

# Beide Kurven zeigen dasselbe Signal auf unterschiedlich feinem Raster.
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(high_time, high_signal, label="fein: 80 Hz", color="tab:blue")
ax.plot(low_time, low_signal, "o--", label="grob: 8 Hz", color="tab:orange")

# Beschriftungen erklären die Einheiten der Zeitachse.
ax.set_title("Zeitachse und Abtastung eines einfachen Signals")
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Signalwert")
ax.legend()

plt.show()

```

### **1.2. Sinus mit Rauschen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_01_02.jpg?v=1784787031" width="250">



>* Sinus beschreibt regelmäßige, wiederkehrende Schwingungen
>* Zeitachse macht Messwerte als Zeitreihe verständlich

>* Rauschen verursacht kleine zufällige Messabweichungen
>* Verrauschte Sinusse trainieren Mustererkennung

>* Kurven zeigen Rauschen, Regelmäßigkeit und Abtastung
>* Künstliche Signale helfen kontrolliert üben



```
#@title Python-Code - Sinus mit Rauschen

# Wir erzeugen ein einfaches Sinussignal.
# Rauschen macht die Messreihe realistischer.
# Die Grafik zeigt beide Verläufe gemeinsam.

import numpy as np
import matplotlib.pyplot as plt

# Eine feste Zufallsquelle macht das Ergebnis reproduzierbar.
rng = np.random.default_rng(42)

# Die Zeitachse beschreibt fünf Sekunden Messdauer.
sampling_rate_hz = 100
duration_seconds = 5
sample_count = sampling_rate_hz * duration_seconds

# Gleichmäßig verteilte Zeitpunkte bilden die Messzeitpunkte.
time_seconds = np.linspace(0, duration_seconds, sample_count, endpoint=False)

# Frequenz und Amplitude bestimmen die ideale Sinusform.
frequency_hz = 2
amplitude = 1.0
clean_signal = amplitude * np.sin(2 * np.pi * frequency_hz * time_seconds)

# Zufälliges Rauschen wird zu jedem Messpunkt addiert.
noise_strength = 0.25
noise = rng.normal(loc=0.0, scale=noise_strength, size=sample_count)
noisy_signal = clean_signal + noise

# Eine einfache Prüfung schützt vor unerwarteten Längenfehlern.
if len(time_seconds) != len(noisy_signal):
    raise ValueError("Zeitachse und Signal müssen gleich lang sein.")

# Kurze Kennzahlen helfen beim Einordnen des Rauschens.
print(f"Messpunkte: {sample_count}")
print(f"Abtastrate: {sampling_rate_hz} Hz")
print(f"Rauschstärke: {noise_strength}")

# Eine gemeinsame Kurve zeigt Muster und Störung gleichzeitig.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(time_seconds, clean_signal, label="Idealer Sinus", linewidth=2)
ax.plot(time_seconds, noisy_signal, label="Sinus mit Rauschen", alpha=0.75)

# Beschriftungen machen die Zeitreihe interpretierbar.
ax.set_title("Sinussignal mit zufälligem Rauschen")
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Signalwert")
ax.legend()

plt.show()

```

### **1.3. CSV Messreihe laden**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_01_03.jpg?v=1784787033" width="250">



>* CSV-Dateien enthalten Zeit- und Messwertspalten
>* Zeitachse prüfen oder aus Abtastrate ableiten

>* CSV-Daten auf Vollständigkeit und Format prüfen
>* Ausreißer, Zeitspalten und Wertebereiche kontrollieren

>* Zeitreihe macht Signalstrukturen sichtbar
>* Visualisierung bereitet spätere Merkmalsanalyse vor



```
#@title Python-Code - CSV Messreihe laden

# Wir laden eine CSV-Messreihe aus Text.
# Die Zeitachse macht das Signal interpretierbar.
# Die Grafik zeigt Trend und Rauschen.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Diese Zeichenkette simuliert eine kleine CSV-Datei.
csv_text = """time_s,temperature_c
0.0,20.1
0.5,20.4
1.0,20.6
1.5,20.9
2.0,21.1
2.5,21.0
3.0,21.4
3.5,21.7
4.0,21.9
4.5,22.1
5.0,22.0
5.5,22.4
6.0,22.7
"""

# Pandas liest hier CSV-Inhalt direkt aus einer Textvariable.
rows = []
for line in csv_text.strip().splitlines()[1:]:
    parts = line.split(",")
    rows.append((float(parts[0]), float(parts[1])))

# Aus den gelesenen Zeilen entsteht eine übersichtliche Tabelle.
data = pd.DataFrame(rows, columns=["time_s", "temperature_c"])

# Eine einfache Prüfung schützt vor falsch geladenen Messreihen.
if len(data) < 2:
    raise ValueError("Die Messreihe braucht mindestens zwei Messpunkte.")

# Die Abtastabstände zeigen, ob die Zeitachse gleichmäßig ist.
time_steps = np.diff(data["time_s"].to_numpy())
mean_step = float(np.mean(time_steps))

# Kurze Ausgaben helfen bei der ersten Sichtprüfung.
print(f"Geladene Messpunkte: {len(data)}")
print(f"Mittlerer Zeitabstand: {mean_step:.2f} s")
print(f"Temperaturbereich: {data['temperature_c'].min():.1f} bis {data['temperature_c'].max():.1f} °C")

# Die Visualisierung verbindet Tabelle und Signalverlauf.
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(data["time_s"], data["temperature_c"], marker="o", label="Messwerte")

# Achsenbeschriftungen machen Einheiten und Bedeutung klar.
ax.set_title("Geladene CSV-Messreihe als Zeitreihe")
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Temperatur in °C")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()

```

## **2. Fenster und Merkmale**

### **2.1. Gleitende Fenster**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_02_01.jpg?v=1784787043" width="250">



>* Signale in kurze Zeitabschnitte zerlegen
>* Lokale Veränderungen sichtbar und messbar machen

>* Fensterlänge bestimmt Detailgrad und Stabilität
>* Schrittweite steuert Auflösung und Aufwand

>* Fenster machen Messreihen lokal auswertbar
>* Merkmale verbinden Rohdaten mit Bedeutung



```
#@title Python-Code - Gleitende Fenster

# Dieses Beispiel zeigt gleitende Fenster.
# Lokale Merkmale beschreiben kurze Signalabschnitte.
# Die Grafik macht Veränderungen sichtbar.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines Signal mit Trend und Rauschen.
rng = np.random.default_rng(42)
time_seconds = np.linspace(0, 20, 201)
trend = 0.04 * time_seconds

# Eine kurze Aktivitätsphase erhöht die Signalwerte lokal.
activity = np.where((time_seconds >= 8) & (time_seconds <= 12), 1.2, 0.0)
noise = rng.normal(0, 0.12, size=time_seconds.size)
signal = np.sin(2 * np.pi * 0.25 * time_seconds) + trend + activity + noise

# Diese Prüfung schützt vor unerwarteten Längenfehlern.
if len(time_seconds) != len(signal):
    raise ValueError("Zeitachse und Signal müssen gleich lang sein.")

# Fensterlänge und Schrittweite bestimmen die lokale Auflösung.
window_size = 25
step_size = 10
window_rows = []

# Jedes Fenster liefert einen Mittelpunkt und einfache Merkmale.
for start in range(0, len(signal) - window_size + 1, step_size):
    end = start + window_size
    window = signal[start:end]
    center_time = time_seconds[start:end].mean()
    mean_value = window.mean()
    energy_value = np.mean(window ** 2)
    window_rows.append([center_time, mean_value, energy_value])

# Aus den Fensterwerten entsteht eine kleine Merkmalstabelle.
features = pd.DataFrame(
    window_rows,
    columns=["Zeitmitte_s", "Mittelwert", "Energie"]
)

# Wir zeigen nur wenige Zeilen, damit die Ausgabe übersichtlich bleibt.
print("Erste Fenstermerkmale:")
print(features.head(4).round(3).to_string(index=False))

# Die Grafik verbindet Rohsignal und lokale Mittelwerte.
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(time_seconds, signal, label="Signal", color="steelblue", alpha=0.65)
ax.plot(features["Zeitmitte_s"], features["Mittelwert"], label="Fenster-Mittelwert", color="darkorange", marker="o")
ax.set_title("Gleitende Fenster: lokale Mittelwerte im Signal")

# Achsenbeschriftungen nennen die verwendeten Einheiten.
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Signalwert")
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()

```

### **2.2. Lokale Kennzahlen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_02_02.jpg?v=1784787045" width="250">



>* Kurze Fenster zeigen lokales Signalverhalten
>* Mittelwert und Streuung machen Veränderungen sichtbar

>* Mittelwert, Median, Streuung und Extremwerte nutzen
>* Kennzahlen gemeinsam statt einzeln bewerten

>* Fensterlänge steuert Reaktion und Glättung
>* Merkmale müssen zur Signalfrage passen



```
#@title Python-Code - Lokale Kennzahlen

# Dieses Beispiel berechnet lokale Kennzahlen aus Signalfenstern.
# Fenster zeigen Mittelwert, Streuung und Extremwerte.
# Die Grafik macht lokale Veränderungen sichtbar.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Wir erzeugen ein kleines Sensorsignal mit Trend und Rauschen.
rng = np.random.default_rng(42)
time_seconds = np.arange(0, 60, 1)
trend = 0.03 * time_seconds
wave = 0.8 * np.sin(2 * np.pi * time_seconds / 20)
noise = rng.normal(0, 0.25, size=time_seconds.size)
signal = 20 + trend + wave + noise

# Eine kurze Störung macht lokale Kennzahlen besonders anschaulich.
signal[32:37] = signal[32:37] + 1.8
window_size = 10
step_size = 5

# Diese Prüfung verhindert unpassende Fenstergrößen.
if window_size > signal.size:
    raise ValueError("Die Fensterlänge muss kleiner als das Signal sein.")

# Für jedes Fenster berechnen wir einfache lokale Kennzahlen.
rows = []
for start in range(0, signal.size - window_size + 1, step_size):
    end = start + window_size
    window = signal[start:end]
    rows.append(
        {
            "start_s": start,
            "mean": np.mean(window),
            "std": np.std(window),
            "min": np.min(window),
            "max": np.max(window),
        }
    )

# Die Tabelle enthält pro Fenster eine kompakte Beschreibung.
features = pd.DataFrame(rows)
print("Erste lokale Kennzahlen je Fenster:")
print(features.head(4).round(2).to_string(index=False))

# Wir zeichnen Signal und lokale Mittelwerte gemeinsam.
window_centers = features["start_s"].to_numpy() + window_size / 2
plt.figure(figsize=(9, 4))
plt.plot(time_seconds, signal, label="Sensorsignal", color="steelblue")
plt.plot(window_centers, features["mean"], "o-", label="lokaler Mittelwert")

# Achsenbeschriftungen nennen die verwendeten Einheiten.
plt.title("Lokale Kennzahlen in gleitenden Fenstern")
plt.xlabel("Zeit in Sekunden")
plt.ylabel("Messwert in °C")
plt.legend()
plt.tight_layout()
plt.show()

```

### **2.3. Änderungen erkennen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_02_03.jpg?v=1784787041" width="250">



>* Signalverlauf wichtiger als Einzelwert
>* Änderungen zeigen Ereignisse und Dynamik

>* Benachbarte Werte zeigen lokale Signaländerungen
>* Fenster helfen, Rauschen von Trends zu trennen

>* Signalabschnitte zeigen Trends, Sprünge und Richtungswechsel
>* Änderungen verbinden Messwerte mit verständlichen Ereignissen



```
#@title Python-Code - Änderungen erkennen

# Dieses Beispiel erkennt Änderungen in einem Signal.
# Differenzen zeigen schnelle Anstiege und Abfälle.
# Die Grafik markiert auffällige Änderungsstellen.

import numpy as np
import matplotlib.pyplot as plt

# Wir erzeugen eine kleine Messreihe mit Zeitachse.
rng = np.random.default_rng(42)
time_seconds = np.arange(0, 60)

# Das Signal enthält Trend, Rauschen und einen Sprung.
trend = 0.03 * time_seconds
noise = rng.normal(0, 0.12, size=time_seconds.size)
signal = 20 + trend + noise
signal[30:] = signal[30:] + 2.0

# Die erste Differenz misst Änderung pro Sekunde.
change = np.diff(signal)
change_time = time_seconds[1:]

# Ein einfacher Schwellenwert findet starke Änderungen.
threshold = 0.8
strong_change_mask = np.abs(change) > threshold
strong_change_times = change_time[strong_change_mask]

# Eine kurze Prüfung macht die Annahmen sichtbar.
if signal.size != time_seconds.size:
    raise ValueError("Signal und Zeitachse müssen gleich lang sein.")

# Wir geben wenige Kennzahlen verständlich aus.
print(f"Anzahl Messpunkte: {signal.size}")
print(f"Größte Änderung: {np.max(np.abs(change)):.2f} Einheiten pro Sekunde")
print(f"Erkannte starke Änderungen: {strong_change_times.size}")
print(f"Zeitpunkte starker Änderungen: {strong_change_times.tolist()}")

# Die Grafik zeigt Signal und erkannte Änderungsstellen.
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(time_seconds, signal, label="Signal", color="steelblue")
ax.scatter(strong_change_times, signal[strong_change_times], color="crimson", label="starke Änderung")
ax.set_title("Änderungen in einer Messreihe erkennen")
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Messwert in Einheiten")
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()

```

## **3. Frequenzen und Merkmalstabellen**

### **3.1. Fehlwerte und Ausreißer**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_03_01.jpg?v=1784787035" width="250">



>* Fehlwerte und Ausreißer verzerren Signalmerkmale
>* Abschnitte vor der Merkmalstabelle prüfen

>* Fehlwerte beeinflussen spätere Signalmerkmale stark
>* Qualitätsmerkmale machen Datenlücken nachvollziehbar

>* Ausreißer kontextabhängig prüfen und behandeln
>* Merkmalstabellen mit Zuverlässigkeitshinweisen ergänzen



```
#@title Python-Code - Fehlwerte und Ausreißer

# Wir untersuchen Fehlwerte in kurzen Signalabschnitten.
# Qualitätsmerkmale ergänzen die berechneten Signalmerkmale.
# Die Tabelle zeigt bereinigte und robuste Kennzahlen.

import numpy as np
import pandas as pd

# Ein kleines Signal erhält Trend, Rauschen und Qualitätsprobleme.
rng = np.random.default_rng(42)
time_seconds = np.arange(0, 12, 0.5)
clean_signal = 0.2 * time_seconds + np.sin(2 * np.pi * time_seconds / 4)
signal = clean_signal + rng.normal(0, 0.12, size=time_seconds.size)

# Fehlwerte und Ausreißer werden gezielt eingefügt.
signal[[5, 6, 15]] = np.nan
signal[[10, 20]] = [4.8, -3.2]
window_id = np.repeat(np.arange(1, 5), 6)

# Die Messreihe wird als Tabelle mit Zeitfenstern gespeichert.
data = pd.DataFrame(
    {"window": window_id, "time_s": time_seconds, "value": signal}
)

# Fehlwerte werden innerhalb der Zeitreihe linear interpoliert.
data["filled_value"] = data["value"].interpolate(limit_direction="both")

# Ausreißer werden mit einer robusten Median-Regel markiert.
median_value = data["filled_value"].median()
mad_value = np.median(np.abs(data["filled_value"] - median_value))
robust_z = 0.6745 * (data["filled_value"] - median_value) / mad_value
data["is_outlier"] = np.abs(robust_z) > 3.5

# Für robuste Merkmale werden extreme Werte begrenzt.
lower_limit = data["filled_value"].quantile(0.05)
upper_limit = data["filled_value"].quantile(0.95)
data["clean_value"] = data["filled_value"].clip(lower_limit, upper_limit)

# Pro Abschnitt entstehen Signalmerkmale und Qualitätsmerkmale.
feature_table = data.groupby("window").agg(
    mean_value=("clean_value", "mean"),
    std_value=("clean_value", "std"),
    missing_share=("value", lambda values: values.isna().mean()),
    outlier_count=("is_outlier", "sum"),
)

# Die Ausgabe bleibt klein und zeigt den Nutzen der Qualitätsmerkmale.
feature_table = feature_table.round(2)
print("Merkmalstabelle pro Signalabschnitt:")
print(feature_table.to_string())

```

### **3.2. Frequenzen sichtbar machen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_03_02.jpg?v=1784787037" width="250">



>* Frequenzen zeigen wiederkehrende Muster im Signal
>* Frequenzmerkmale machen Signalabschnitte vergleichbar

>* Kurze Fenster zeigen veränderte Frequenzmuster.
>* Einheitliche Merkmale füllen vergleichbare Tabellen.

>* Frequenzen als verständliche Kennwerte zusammenfassen
>* Kontext macht Merkmale vergleichbar und nutzbar



```
#@title Python-Code - Frequenzen sichtbar machen

# Wir machen Frequenzen in Signalabschnitten sichtbar.
# Die Fourier-Analyse zeigt wiederkehrende Schwingungen.
# Eine kleine Tabelle fasst Abschnittsmerkmale zusammen.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Wir erzeugen ein kurzes Signal mit zwei Abschnitten.
sampling_rate = 100
seconds = 4
time = np.arange(0, seconds, 1 / sampling_rate)

# Der erste Abschnitt schwingt langsamer als der zweite.
first_part = np.sin(2 * np.pi * 3 * time[:200])
second_part = np.sin(2 * np.pi * 8 * time[200:])
signal = np.concatenate([first_part, second_part])

# Ein kleines Rauschen macht das Beispiel realistischer.
rng = np.random.default_rng(42)
noise = rng.normal(0, 0.25, size=signal.size)
signal = signal + noise

# Wir prüfen, ob die Fenster genau passen.
window_size = 200
if signal.size % window_size != 0:
    raise ValueError("Die Signalabschnitte passen nicht genau.")

# Für jedes Fenster berechnen wir Frequenzmerkmale.
rows = []
for start in range(0, signal.size, window_size):
    window = signal[start:start + window_size]
    frequencies = np.fft.rfftfreq(window_size, d=1 / sampling_rate)
    amplitudes = np.abs(np.fft.rfft(window))
    amplitudes[0] = 0
    dominant_index = np.argmax(amplitudes)
    dominant_frequency = frequencies[dominant_index]
    low_band_strength = amplitudes[(frequencies >= 1) & (frequencies <= 5)].sum()
    high_band_strength = amplitudes[(frequencies > 5) & (frequencies <= 12)].sum()
    rows.append([start // window_size + 1, dominant_frequency, low_band_strength, high_band_strength])

# Die Merkmalstabelle enthält eine Zeile pro Abschnitt.
feature_table = pd.DataFrame(
    rows,
    columns=["Abschnitt", "dominante_Frequenz_Hz", "Staerke_1_bis_5_Hz", "Staerke_5_bis_12_Hz"]
)

# Gerundete Werte sind für den Einstieg leichter lesbar.
print("Frequenzmerkmale pro Signalabschnitt:")
print(feature_table.round(2).to_string(index=False))

# Wir zeigen das Frequenzspektrum des zweiten Abschnitts.
window = signal[200:400]
frequencies = np.fft.rfftfreq(window_size, d=1 / sampling_rate)
amplitudes = np.abs(np.fft.rfft(window))

# Die höchste Spitze markiert die dominante Frequenz.
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(frequencies, amplitudes, label="Spektrum Abschnitt 2")
ax.set_title("Frequenzen im zweiten Signalabschnitt")
ax.set_xlabel("Frequenz in Hz")
ax.set_ylabel("Amplitude")
ax.set_xlim(0, 15)
ax.legend()
plt.show()

```

### **3.3. Signalprojekt mit Merkmalen**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/Udemy/ML Für Anfänger/Module_05/Lecture_B/image_03_03.jpg?v=1784787039" width="250">



>* Signalabschnitte werden zu Merkmalen zusammengefasst
>* Merkmalstabellen machen Signale vergleichbar und nutzbar

>* Abschnittslänge und Überlappung bewusst wählen
>* Fehlwerte und Ausreißer vorher bereinigen

>* Merkmale verbinden Rohsignale mit Fachfragen
>* Klare Metadaten machen Tabellen zuverlässig



```
#@title Python-Code - Signalprojekt mit Merkmalen

# Dieses Mini-Projekt erzeugt ein künstliches Messsignal.
# Signalabschnitte werden zu verständlichen Merkmalen verdichtet.
# Am Ende entsteht eine kleine Merkmalstabelle.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Eine feste Zufallsquelle macht das Beispiel wiederholbar.
rng = np.random.default_rng(42)

# Wir erzeugen vier Sekunden Messwerte mit 50 Hertz.
sampling_rate_hz = 50
duration_seconds = 4
sample_count = sampling_rate_hz * duration_seconds

# Die Zeitachse beschreibt jeden Messpunkt in Sekunden.
time_seconds = np.arange(sample_count) / sampling_rate_hz

# Das Signal enthält Trend, Schwingung und kleines Rauschen.
trend = 0.15 * time_seconds
wave = np.sin(2 * np.pi * 3 * time_seconds)
noise = rng.normal(0, 0.18, sample_count)
signal = trend + wave + noise

# Eine einfache Prüfung schützt vor falschen Abschnittsgrößen.
window_size = sampling_rate_hz
if sample_count % window_size != 0:
    raise ValueError("Die Messreihe passt nicht exakt in ganze Sekunden.")

# Jeder Abschnitt wird als eigene Beobachtung betrachtet.
windows = signal.reshape(-1, window_size)
window_count = windows.shape[0]

# Änderungswerte zeigen, wie stark das Signal springt.
changes = np.diff(windows, axis=1)

# Die Frequenzachse gehört zur schnellen Fourier-Transformation.
frequencies_hz = np.fft.rfftfreq(window_size, d=1 / sampling_rate_hz)
spectra = np.abs(np.fft.rfft(windows, axis=1))

# Die Nullfrequenz wird ignoriert, damit Schwingungen dominieren.
dominant_indices = np.argmax(spectra[:, 1:], axis=1) + 1
dominant_frequencies_hz = frequencies_hz[dominant_indices]

# Aus jedem Abschnitt entsteht eine Zeile der Merkmalstabelle.
feature_table = pd.DataFrame({
    "abschnitt": np.arange(1, window_count + 1),
    "start_s": np.arange(window_count),
    "mittelwert": windows.mean(axis=1),
    "streuung": windows.std(axis=1),
    "dominante_hz": dominant_frequencies_hz,
})

# Gerundete Werte sind für den ersten Blick leichter lesbar.
print("Merkmalstabelle: eine Zeile beschreibt eine Signalsekunde.")
print(feature_table.round(2).to_string(index=False))

# Die Grafik zeigt Rohsignal und Abschnittsgrenzen gemeinsam.
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(time_seconds, signal, label="Messsignal")

# Senkrechte Linien markieren die gebildeten Abschnitte.
for boundary in range(1, duration_seconds):
    ax.axvline(boundary, color="gray", linestyle="--", linewidth=1)

# Beschriftungen machen die Darstellung interpretierbar.
ax.set_title("Signalabschnitte als Grundlage einer Merkmalstabelle")
ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel("Signalwert")
ax.legend()
plt.show()

```
