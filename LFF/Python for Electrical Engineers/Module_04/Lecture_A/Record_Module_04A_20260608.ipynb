## **1. Resistance Networks**

### **1.1. Series Resistance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1780963527" width="250">



>* Series resistors share one current path
>* Total resistance equals all resistances added

>* Add each resistor to find total resistance
>* Series calculations are simple and checkable

>* Use consistent resistance units before computing.
>* Relate results to real circuit limits.



```
#@title Python Code - Series Resistance

# Compute series resistance using simple Python functions.
# Values must share the same resistance unit.
# This example also checks a design requirement.

# Define a reusable function for series resistors.
def series_resistance(resistor_values):
    """Return equivalent resistance for a series network."""
    if not resistor_values:
        raise ValueError("Provide at least one resistor value.")

    # Reject invalid resistance values before summing.
    for resistance in resistor_values:
        if resistance < 0:
            raise ValueError("Resistance values cannot be negative.")

    # Series resistance is the sum of all resistors.
    total_resistance = sum(resistor_values)
    return total_resistance

# Store example resistor values in ohms.
resistors_ohms = [220, 330, 470]
supply_voltage = 9.0

# Compute equivalent resistance and circuit current.
total_ohms = series_resistance(resistors_ohms)
circuit_current = supply_voltage / total_ohms

# Compute power dissipated by each resistor.
powers_watts = []
for resistance in resistors_ohms:
    powers_watts.append(circuit_current**2 * resistance)

# Check one simple current design requirement.
current_limit = 0.020
design_ok = circuit_current <= current_limit

# Print concise results for the learner.
print("Series resistors (ohms):", resistors_ohms)
print("Equivalent resistance:", total_ohms, "ohms")
print("Circuit current:", round(circuit_current * 1000, 2), "mA")
print("Power per resistor:", [round(power, 4) for power in powers_watts])
print("Current requirement met:", design_ok)

```

### **1.2. Parallel Resistance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1780963525" width="250">



>* Parallel branches share voltage and split current
>* Python must sum conductance, not resistance

>* Parallel branches share voltage and divide current
>* Python computes equivalent total supply load

>* Low-resistance branches can dominate parallel networks
>* Check Python results against circuit meaning



```
#@title Python Code - Parallel Resistance

# Parallel branches share voltage across every resistor.
# Conductance values add before converting back.
# A short circuit makes resistance very small.

def parallel_resistance(resistors_ohms):
    # Reject empty resistor lists before computing.
    if len(resistors_ohms) == 0:
        raise ValueError("Provide at least one resistor.")

    # Check every resistance is physically valid.
    for resistance in resistors_ohms:
        if resistance <= 0:
            raise ValueError("Resistances must be positive.")

    # Add branch conductances for parallel paths.
    total_conductance = sum(1 / resistance for resistance in resistors_ohms)
    equivalent_resistance = 1 / total_conductance
    return equivalent_resistance

# Define a small parallel resistor network.
branches_ohms = [100, 220, 470]
supply_voltage = 12
smallest_branch = min(branches_ohms)

# Compute equivalent resistance for all branches.
req_ohms = parallel_resistance(branches_ohms)
total_current = supply_voltage / req_ohms
branch_currents = [supply_voltage / r for r in branches_ohms]

# Compute power dissipated in each branch.
branch_powers = [supply_voltage * current for current in branch_currents]
rating_watts = 0.5
safe_flags = [power <= rating_watts for power in branch_powers]

# Print compact results for circuit interpretation.
print(f"Parallel branches: {branches_ohms} ohms")
print(f"Equivalent resistance: {req_ohms:.2f} ohms")
print(f"Smallest branch resistance: {smallest_branch:.2f} ohms")
print(f"Total source current: {total_current:.3f} A")
print(f"Branch currents: {[round(i, 3) for i in branch_currents]} A")
print(f"Branch powers: {[round(p, 3) for p in branch_powers]} W")
print(f"All branches under {rating_watts} W rating: {all(safe_flags)}")

```

### **1.3. Mixed Networks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1780963529" width="250">



>* Mixed networks need step-by-step simplification
>* Python functions mirror series-parallel calculations

>* Simplify local series or parallel groups first
>* Use Python functions to combine equivalents

>* Connection patterns determine equivalent resistance
>* Use Python to simplify and verify networks



```
#@title Python Code - Mixed Networks

# Mixed networks combine simpler resistor groups.
# Python functions mirror stepwise circuit simplification.
# Ratings checks connect computation to design.

# Define a safe series resistance function.
def series_resistance(resistor_values):
    values = list(resistor_values)
    assert len(values) > 0, "Provide at least one resistor."

    assert all(value > 0 for value in values)
    return sum(values)

# Define a safe parallel resistance function.
def parallel_resistance(resistor_values):
    values = list(resistor_values)
    assert len(values) > 0, "Provide at least one resistor."

    assert all(value > 0 for value in values)
    conductance = sum(1 / value for value in values)
    return 1 / conductance

# Store resistor values in ohms.
r1_ohm = 220
r2_ohm = 330
r3_ohm = 470

# Simplify the deepest parallel pair first.
parallel_pair_ohm = parallel_resistance([r2_ohm, r3_ohm])
total_ohm = series_resistance([r1_ohm, parallel_pair_ohm])
source_voltage = 12.0

# Compute total current from Ohm's law.
total_current_amp = source_voltage / total_ohm
series_power_watt = total_current_amp**2 * r1_ohm
parallel_voltage = source_voltage - total_current_amp * r1_ohm

# Compute branch currents and branch powers.
r2_current_amp = parallel_voltage / r2_ohm
r3_current_amp = parallel_voltage / r3_ohm
r2_power_watt = parallel_voltage * r2_current_amp
r3_power_watt = parallel_voltage * r3_current_amp

# Check a common quarter-watt rating.
rating_watt = 0.25
power_values = [series_power_watt, r2_power_watt, r3_power_watt]
rating_ok = all(power < rating_watt for power in power_values)

# Print a compact analysis summary.
print(f"Parallel pair: {parallel_pair_ohm:.2f} ohms")
print(f"Total equivalent: {total_ohm:.2f} ohms")
print(f"Source current: {1000 * total_current_amp:.2f} mA")
print(f"Parallel voltage: {parallel_voltage:.2f} V")
print(f"Branch currents: {1000*r2_current_amp:.2f}, {1000*r3_current_amp:.2f} mA")
print(f"Powers: {series_power_watt:.3f}, {r2_power_watt:.3f}, {r3_power_watt:.3f} W")
print(f"All resistors below {rating_watt:.2f} W rating: {rating_ok}")

```

## **2. DC Operating Points**

### **2.1. Node Voltages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1780963537" width="250">



>* Nodes share voltage along connected conductors
>* Reference ground defines all node voltages

>* Voltage differences drive component currents
>* Node voltages verify circuit signal levels

>* Check reference node and model assumptions
>* Real wiring can change circuit voltages



```
#@title Python Code - Node Voltages

# Compute node voltages using a divider.
# Compare each node against ground.
# Use voltages to find current.

# Define the DC supply voltage.
supply_voltage = 12.0

# Define two series resistor values.
resistor_top = 1000.0
resistor_bottom = 2000.0

# Calculate the total series resistance.
total_resistance = resistor_top + resistor_bottom

# Calculate the same branch current.
branch_current = supply_voltage / total_resistance

# Calculate the middle node voltage.
node_ground = 0.0
node_supply = supply_voltage
node_middle = branch_current * resistor_bottom

# Calculate voltage drops across resistors.
top_drop = node_supply - node_middle
bottom_drop = node_middle - node_ground

# Calculate power dissipated in each resistor.
top_power = branch_current**2 * resistor_top
bottom_power = branch_current**2 * resistor_bottom

# Check a simple component rating.
rating_watts = 0.25
safe_top = top_power < rating_watts
safe_bottom = bottom_power < rating_watts

# Print a compact operating point summary.
print(f"Supply node: {node_supply:.2f} V")
print(f"Middle node: {node_middle:.2f} V")
print(f"Ground node: {node_ground:.2f} V")
print(f"Branch current: {branch_current*1000:.2f} mA")
print(f"Top resistor drop: {top_drop:.2f} V")
print(f"Bottom resistor drop: {bottom_drop:.2f} V")
print(f"Top power: {top_power*1000:.1f} mW, safe: {safe_top}")
print(f"Bottom power: {bottom_power*1000:.1f} mW, safe: {safe_bottom}")

```

### **2.2. Branch Currents**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1780963539" width="250">



>* Current depends on node voltage differences
>* Direction signs show actual branch flow

>* Parallel branches share voltage, currents vary
>* Series components all carry identical current

>* Check current limits to prevent circuit damage
>* Use branch currents to verify safe designs



```
#@title Python Code - Branch Currents

# Calculate branch currents using node voltage differences.
# Positive current follows the chosen reference direction.
# Compare currents with simple component current ratings.

# Store node voltages relative to circuit ground.
node_voltages = {"VIN": 12.0, "N1": 5.0, "GND": 0.0}

# Define branches as from node, to node, resistance, limit.
branches = [
    ("R1", "VIN", "N1", 700.0, 0.020),
    ("R2", "N1", "GND", 1000.0, 0.010),
    ("LED", "N1", "GND", 2500.0, 0.003),
]

# Compute current from voltage difference and resistance.
def branch_current(from_voltage, to_voltage, resistance_ohms):
    return (from_voltage - to_voltage) / resistance_ohms

# Print a compact table header for learners.
print("Branch  Current_mA  Power_mW  Status")

# Analyze each branch using its reference direction.
for name, start, end, resistance, limit in branches:
    current = branch_current(node_voltages[start], node_voltages[end], resistance)
    power = (current ** 2) * resistance
    status = "OK" if abs(current) <= limit else "OVER LIMIT"

    # Display signed current and dissipated power.
    print(f"{name:6s} {current*1000:9.2f} {power*1000:8.2f} {status}")

# Check current balance at node N1.
inflow = branch_current(node_voltages["VIN"], node_voltages["N1"], 700.0)
outflow_r2 = branch_current(node_voltages["N1"], node_voltages["GND"], 1000.0)
outflow_led = branch_current(node_voltages["N1"], node_voltages["GND"], 2500.0)

# A small mismatch suggests the node voltage needs refinement.
mismatch = inflow - outflow_r2 - outflow_led
print(f"Node N1 current mismatch = {mismatch*1000:.2f} mA")

```

### **2.3. Power Dissipation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1780963541" width="250">



>* Power converts electrical energy into other forms
>* Voltage and current determine energy transfer rate

>* Power reveals unsafe overheating risks
>* Wasted heat reduces efficiency and reliability

>* Check source and load energy balance
>* Use power to choose reliable components



```
#@title Python Code - Power Dissipation

# Compute resistor power from operating points.
# Compare results with component ratings.
# Keep outputs short for beginners.

# No extra packages are needed.

def resistor_power(voltage_volts, current_amps):
    return voltage_volts * current_amps

# Store known branch operating points.
resistors = [
    ("R1", 9.0, 0.006, 0.25),
    ("R2", 3.0, 0.006, 0.125),
    ("R3", 5.0, 0.020, 0.25),
]

# Print a compact table header.
print("Resistor  Power(W)  Rating(W)  Status")

# Calculate power and compare ratings.
for name, volts, amps, rating in resistors:
    power_watts = resistor_power(volts, amps)
    status = "OK" if power_watts <= rating else "Too hot"
    print(f"{name:>3}      {power_watts:6.3f}     {rating:6.3f}   {status}")

# Check total power absorbed by resistors.
total_power = sum(resistor_power(v, i) for _, v, i, _ in resistors)

# Compare source delivery with load absorption.
source_power = 12.0 * 0.026
print(f"Total resistor power: {total_power:.3f} W")
print(f"Source delivered power: {source_power:.3f} W")

```

## **3. Design Verification**

### **3.1. Rating Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1780963530" width="250">



>* Compare calculated values with component limits
>* Use ratings to judge circuit safety

>* Match each component to its stress type
>* Functional circuits can still exceed safe ratings

>* Check ratings under real operating conditions
>* Use worst cases for reliable designs



```
#@title Python Code - Rating Checks

# Verify component ratings after circuit calculations.
# Compare computed stress against safe limits.
# Print concise pass fail design results.

# Store circuit and component assumptions.
supply_v = 12.0
resistance_ohm = 470.0
power_rating_w = 0.25

# Store additional device rating limits.
switch_rating_a = 0.050
led_forward_v = 2.0
required_current_a = 0.020

# Calculate branch current through series path.
resistor_voltage_v = supply_v - led_forward_v
branch_current_a = resistor_voltage_v / resistance_ohm
resistor_power_w = branch_current_a**2 * resistance_ohm

# Define a reusable rating check function.
def check_limit(name, value, limit, unit):
    margin = limit - value
    status = "PASS" if value <= limit else "FAIL"
    return f"{name}: {value:.3f} {unit} <= {limit:.3f} {unit} -> {status}"

# Check whether current also meets requirement.
current_ok = branch_current_a >= required_current_a
current_status = "PASS" if current_ok else "FAIL"

# Print a compact design verification summary.
print("DC indicator circuit rating checks")
print(check_limit("Resistor power", resistor_power_w, power_rating_w, "W"))
print(check_limit("Switch current", branch_current_a, switch_rating_a, "A"))
print(f"LED current requirement: {branch_current_a:.3f} A >= {required_current_a:.3f} A -> {current_status}")

# Summarize the final design decision.
all_checks_ok = (resistor_power_w <= power_rating_w) and (branch_current_a <= switch_rating_a) and current_ok
final_status = "ACCEPT DESIGN" if all_checks_ok else "CHANGE COMPONENTS"
print(f"Final decision: {final_status}")

```

### **3.2. Component Rating Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1780963532" width="250">



>* Compare calculated values with component limits
>* Ensure parts operate safely and reliably

>* Match worst stresses to datasheet limits
>* Check parts under realistic operating conditions

>* Consider variation, environment, and long-term reliability
>* Revise designs to ensure safe durable operation



```
#@title Python Code - Component Rating Checks

# Verify component ratings using calculated circuit values.
# Compare operating values against safe derated limits.
# This example uses a simple LED circuit.

# Store supply and component assumptions.
supply_voltage = 12.0
led_forward_voltage = 2.0
series_resistance = 470.0

# Store published component ratings.
resistor_power_rating = 0.25
led_current_rating = 0.020
safety_factor = 0.80

# Compute branch current and resistor power.
resistor_voltage = supply_voltage - led_forward_voltage
branch_current = resistor_voltage / series_resistance
resistor_power = branch_current ** 2 * series_resistance

# Compute conservative allowed operating limits.
safe_resistor_power = resistor_power_rating * safety_factor
safe_led_current = led_current_rating * safety_factor

# Check each component against derated limits.
resistor_ok = resistor_power <= safe_resistor_power
led_ok = branch_current <= safe_led_current
design_ok = resistor_ok and led_ok

# Build compact status words for display.
resistor_status = "PASS" if resistor_ok else "FAIL"
led_status = "PASS" if led_ok else "FAIL"
design_status = "PASS" if design_ok else "FAIL"

# Print calculated operating values.
print(f"LED branch current: {branch_current * 1000:.2f} mA")
print(f"Resistor power: {resistor_power:.3f} W")
print(f"Safe resistor limit: {safe_resistor_power:.3f} W")
print(f"Safe LED current limit: {safe_led_current * 1000:.2f} mA")

# Print rating check results.
print(f"Resistor rating check: {resistor_status}")
print(f"LED rating check: {led_status}")
print(f"Overall design verification: {design_status}")
print("Choose higher ratings or adjust resistance if any check fails.")

```

### **3.3. Design Margin Check**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1780963534" width="250">



>* Compare calculated values with component limits
>* Leave margin for real-world variations

>* Leave power and voltage safety margins
>* Account for transients, batteries, and harsh environments

>* Check performance under worst-case component tolerances
>* Keep operating values safely within limits



```
#@title Python Code - Design Margin Check

# Design margins protect real circuit components.
# Calculated values must stay below ratings.
# This example checks one resistor divider.

# Define source voltage and resistor values.
source_voltage_volts = 12.0
resistor_top_ohms = 1000.0
resistor_bottom_ohms = 2000.0

# Define component ratings and required margin.
power_rating_watts = 0.25
voltage_rating_volts = 16.0
minimum_margin_percent = 50.0

# Compute divider current and node voltage.
total_resistance_ohms = resistor_top_ohms + resistor_bottom_ohms
divider_current_amps = source_voltage_volts / total_resistance_ohms
node_voltage_volts = divider_current_amps * resistor_bottom_ohms

# Compute resistor power dissipation values.
top_power_watts = divider_current_amps**2 * resistor_top_ohms
bottom_power_watts = divider_current_amps**2 * resistor_bottom_ohms
worst_power_watts = max(top_power_watts, bottom_power_watts)

# Define a reusable margin calculation function.
def margin_percent(rating_value, operating_value):
    return (rating_value / operating_value - 1.0) * 100.0

# Calculate design margins for key limits.
power_margin_percent = margin_percent(power_rating_watts, worst_power_watts)
voltage_margin_percent = margin_percent(voltage_rating_volts, source_voltage_volts)
passes_power_margin = power_margin_percent >= minimum_margin_percent

# Check whether voltage margin also passes.
passes_voltage_margin = voltage_margin_percent >= minimum_margin_percent
overall_pass = passes_power_margin and passes_voltage_margin
status_text = "PASS" if overall_pass else "REVISE"

# Print a compact verification summary.
print(f"Node voltage: {node_voltage_volts:.2f} V")
print(f"Divider current: {divider_current_amps*1000:.2f} mA")
print(f"Worst resistor power: {worst_power_watts*1000:.1f} mW")
print(f"Power margin: {power_margin_percent:.0f}%")
print(f"Voltage margin: {voltage_margin_percent:.0f}%")
print(f"Design margin check: {status_text}")

```
