## **1. Complex AC Quantities**

### **1.1. Phasor Representation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1780964516" width="250">



>* Phasors summarize AC amplitude and phase
>* Complex numbers simplify waveform comparison

>* Complex parts show in-phase and quadrature components
>* Phasors simplify adding same-frequency signals

>* Phasors summarize fixed-frequency sinusoidal relationships
>* Phase depends on the chosen reference



```
#@title Python Code - Phasor Representation

# Phasors store sinusoid size and timing.
# Python complex numbers represent phasors directly.
# This example adds two AC phasors.

import cmath
import math
import matplotlib.pyplot as plt

# Define a helper for polar phasors.
def phasor(magnitude_value, angle_degrees):
    angle_radians = math.radians(angle_degrees)
    return magnitude_value * cmath.exp(1j * angle_radians)

# Create two voltage phasors.
voltage_a = phasor(120.0, 0.0)
voltage_b = phasor(80.0, -60.0)

# Add phasors using complex arithmetic.
voltage_sum = voltage_a + voltage_b
phasor_values = [voltage_a, voltage_b, voltage_sum]
phasor_names = ["120 V at 0 deg", "80 V at -60 deg", "Resultant phasor"]

# Print compact rectangular and polar forms.
for name_text, value in zip(phasor_names, phasor_values):
    magnitude_value = abs(value)
    angle_value = math.degrees(cmath.phase(value))
    print(f"{name_text}: {value.real:.1f} + j{value.imag:.1f} V")
    print(f"  magnitude={magnitude_value:.1f} V, phase={angle_value:.1f} deg")

# Prepare coordinates for a phasor diagram.
colors_list = ["tab:blue", "tab:orange", "tab:green"]
fig, ax = plt.subplots(figsize=(6, 6))

# Draw each phasor from the origin.
for value, name_text, color_text in zip(phasor_values, phasor_names, colors_list):
    ax.arrow(0, 0, value.real, value.imag, color=color_text,
             length_includes_head=True, head_width=5.0, head_length=8.0)
    ax.text(value.real * 1.05, value.imag * 1.05, name_text, color=color_text)

# Format the complex plane view.
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)
ax.set_aspect("equal", adjustable="box")
ax.set_xlabel("Real axis: in-phase volts")
ax.set_ylabel("Imaginary axis: quadrature volts")
ax.set_title("Phasor Representation with Python Complex Numbers")
ax.grid(True)
plt.show()

```

### **1.2. Impedance Representation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1780964514" width="250">



>* Impedance extends resistance to AC circuits
>* Complex parts show resistance and timing shift

>* Complex numbers combine resistance and reactance
>* Impedance arithmetic models real AC circuits

>* Reactive impedance changes with signal frequency
>* Complex values simplify AC circuit calculations



```
#@title Python Code - Impedance Representation

# Complex impedance stores resistance and reactance.
# Python complex numbers model AC quantities.
# This example compares simple circuit impedances.

# Define a small example frequency.
frequency_hz = 1000.0
omega_rad_s = 2 * 3.141592653589793 * frequency_hz

# Choose ordinary component values.
resistance_ohm = 100.0
capacitance_f = 1.0e-6
inductance_h = 0.050

# Build impedances using Python complex numbers.
z_resistor = complex(resistance_ohm, 0.0)
z_capacitor = 1 / (1j * omega_rad_s * capacitance_f)
z_inductor = 1j * omega_rad_s * inductance_h

# Combine series elements by adding impedances.
z_series_rc = z_resistor + z_capacitor
z_series_rl = z_resistor + z_inductor

# Create a compact reporting helper.
def describe_impedance(name, z_value):
    magnitude = abs(z_value)
    phase_rad = __import__("cmath").phase(z_value)

    phase_deg = phase_rad * 180.0 / 3.141592653589793
    real_part = z_value.real
    imag_part = z_value.imag

    print(f"{name}: {real_part:.2f} + j{imag_part:.2f} ohms")
    print(f"  magnitude={magnitude:.2f} ohms, phase={phase_deg:.1f} degrees")

# Display each impedance without printing large data.
print(f"Frequency: {frequency_hz:.0f} Hz")
describe_impedance("Resistor", z_resistor)
describe_impedance("Capacitor", z_capacitor)

# Compare combined impedances for common circuits.
describe_impedance("Series RC", z_series_rc)
describe_impedance("Series RL", z_series_rl)

```

### **1.3. Magnitude and Phase**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1780964518" width="250">



>* Magnitude shows AC size or strength
>* Phase shows timing relative to reference

>* Magnitude shows strength or opposition
>* Phase shows lead, lag, or alignment

>* Magnitude shows size; phase shows timing
>* Complex results guide measurements and design



```
#@title Python Code - Magnitude and Phase

# Complex phasors combine size and timing.
# Python complex numbers model AC quantities.
# Magnitude and phase reveal circuit meaning.

import cmath
import math
import matplotlib.pyplot as plt

# Define two voltage phasors with equal RMS magnitude.
voltage_a = 120 * cmath.exp(1j * math.radians(0))
voltage_b = 120 * cmath.exp(1j * math.radians(-45))

# Convert complex phasors into magnitude and angle.
mag_a = abs(voltage_a)
phase_a = math.degrees(cmath.phase(voltage_a))
mag_b = abs(voltage_b)
phase_b = math.degrees(cmath.phase(voltage_b))

# Show rectangular and polar descriptions together.
print(f"Voltage A rectangular: {voltage_a:.2f} V")
print(f"Voltage A magnitude-phase: {mag_a:.1f} V, {phase_a:.1f} degrees")
print(f"Voltage B rectangular: {voltage_b:.2f} V")
print(f"Voltage B magnitude-phase: {mag_b:.1f} V, {phase_b:.1f} degrees")

# Create a simple impedance example.
resistance = 100.0
reactance = 75.0
impedance = complex(resistance, reactance)

# Interpret impedance using magnitude and phase.
imp_mag = abs(impedance)
imp_phase = math.degrees(cmath.phase(impedance))
print(f"Impedance magnitude: {imp_mag:.1f} ohms")
print(f"Impedance phase: {imp_phase:.1f} degrees")

# Draw phasor arrows for visual comparison.
phasors = [voltage_a, voltage_b]
labels = ["A: 0 deg", "B: -45 deg"]
colors = ["tab:blue", "tab:orange"]

# Make one compact phasor plot.
fig, ax = plt.subplots(figsize=(5, 5))
for phasor, label, color in zip(phasors, labels, colors):
    ax.arrow(0, 0, phasor.real, phasor.imag,
             head_width=5, length_includes_head=True,
             color=color, label=label)

# Format axes as real and imaginary components.
ax.set_aspect("equal", adjustable="box")
ax.set_xlim(-20, 140)
ax.set_ylim(-100, 60)
ax.grid(True)

# Label the phasor diagram clearly.
ax.set_xlabel("Real component, volts")
ax.set_ylabel("Imaginary component, volts")
ax.set_title("Equal Magnitude, Different Phase")
ax.legend(loc="lower right")
plt.show()

```

## **2. Reactive Components**

### **2.1. Capacitive Reactance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1780964519" width="250">



>* Capacitors oppose low frequencies more strongly
>* They block DC while passing changing signals

>* Higher frequency or capacitance lowers reactance
>* Capacitor behavior depends on signal frequency

>* Current leads voltage in capacitors
>* Frequency controls reactance and filtering behavior



```
#@title Python Code - Capacitive Reactance

# This example computes capacitive reactance values.
# Frequency strongly changes capacitor opposition.
# Complex impedance includes capacitive phase behavior.

import numpy as np
import matplotlib.pyplot as plt

# Define one practical capacitor value.
capacitance_farads = 1.0e-6
frequencies_hz = np.array([10, 100, 1000, 10000])

# Compute reactance magnitude using the standard formula.
xc_ohms = 1 / (2 * np.pi * frequencies_hz * capacitance_farads)
zc_ohms = -1j * xc_ohms

# Print a compact table of frequency results.
print("Capacitor value: 1.0 microfarad")
print("Frequency (Hz) | Xc magnitude (ohms) | Zc")
for freq, xc, zc in zip(frequencies_hz, xc_ohms, zc_ohms):
    print(f"{freq:12.0f} | {xc:19.2f} | {zc:.2f}")

# Build a smooth frequency range for plotting.
freq_plot_hz = np.logspace(1, 5, 200)
xc_plot_ohms = 1 / (2 * np.pi * freq_plot_hz * capacitance_farads)

# Plot reactance magnitude versus frequency.
plt.figure(figsize=(7, 4))
plt.loglog(freq_plot_hz, xc_plot_ohms, linewidth=2)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Capacitive reactance magnitude (ohms)")

# Add readable plot details.
plt.title("Capacitive Reactance Decreases as Frequency Increases")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

```

### **2.2. Inductive Reactance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780964521" width="250">



>* Inductors resist changing AC current
>* Higher frequency creates greater inductive reactance

>* Inductors resist rapid current changes
>* Used in supplies, crossovers, and motors

>* Resistance and reactance determine impedance magnitude
>* Higher frequency increases lag and filtering effects



```
#@title Python Code - Inductive Reactance

# Explore inductive reactance using Python complex numbers.
# Reactance grows as signal frequency increases.
# Impedance magnitude combines resistance and reactance.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Define circuit values for one coil.
inductance_h = 0.047
resistance_ohm = 12.0
frequencies_hz = np.array([10, 60, 1000, 10000])

# Calculate angular frequency and reactance.
omega_rad_s = 2 * np.pi * frequencies_hz
x_l_ohm = omega_rad_s * inductance_h
z_complex = resistance_ohm + 1j * x_l_ohm

# Calculate impedance magnitude and phase angle.
z_mag_ohm = np.abs(z_complex)
phase_deg = np.degrees(np.angle(z_complex))
print("Inductive reactance example for L = 47 mH.")

# Print compact results for selected frequencies.
for f, xl, zm, ph in zip(frequencies_hz, x_l_ohm, z_mag_ohm, phase_deg):
    print(f"{f:6.0f} Hz: XL={xl:7.2f} ohm, |Z|={zm:7.2f} ohm, phase={ph:5.1f} deg")

# Prepare smooth frequencies for the plot.
plot_frequencies = np.logspace(1, 5, 200)
plot_xl = 2 * np.pi * plot_frequencies * inductance_h
plot_zmag = np.sqrt(resistance_ohm**2 + plot_xl**2)

# Plot reactance and impedance magnitude versus frequency.
plt.figure(figsize=(7, 4))
plt.loglog(plot_frequencies, plot_xl, label="Inductive reactance XL")
plt.loglog(plot_frequencies, plot_zmag, label="Impedance magnitude |Z|")

# Add labels and helpful gridlines.
plt.xlabel("Frequency (Hz)")
plt.ylabel("Ohms")
plt.title("Inductor opposition increases with frequency")
plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Display the final teaching plot.
plt.legend()
plt.tight_layout()
plt.show()

```

### **2.3. Frequency Response**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1780964523" width="250">



>* Reactive impedance changes with signal frequency
>* Capacitors oppose less; inductors oppose more

>* Capacitors pass higher frequencies more easily
>* Inductors increasingly block higher-frequency current

>* Predict which signal frequencies pass or weaken
>* Reactance shapes filtering across frequency



```
#@title Python Code - Frequency Response

# Explore frequency response using reactance.
# Complex impedances represent AC opposition.
# One plot compares capacitor and inductor.

import numpy as np
import matplotlib.pyplot as plt

# Define circuit component values.
R_ohm = 100.0
C_farad = 1.0e-6
L_henry = 10.0e-3

# Sweep frequencies across practical decades.
freq_hz = np.logspace(1, 6, 300)
omega = 2 * np.pi * freq_hz
assert freq_hz.size == omega.size

# Calculate ideal component impedances.
Zc = 1 / (1j * omega * C_farad)
Zl = 1j * omega * L_henry
Zr = R_ohm + 0j

# Calculate magnitudes of reactance.
Xc_mag = np.abs(Zc)
Xl_mag = np.abs(Zl)
Zrc_mag = np.abs(Zr + Zc)

# Inspect three representative frequencies.
sample_freqs = np.array([10.0, 1000.0, 100000.0])
sample_omega = 2 * np.pi * sample_freqs
sample_xc = 1 / (sample_omega * C_farad)
sample_xl = sample_omega * L_henry

# Print compact numerical results.
print("Frequency response sample values:")
for f, xc, xl in zip(sample_freqs, sample_xc, sample_xl):
    print(f"{f:8.0f} Hz: |Xc|={xc:9.1f} ohm, |Xl|={xl:9.1f} ohm")

# Plot frequency dependent opposition.
plt.figure(figsize=(8, 5))
plt.loglog(freq_hz, Xc_mag, label="capacitor |Xc|")
plt.loglog(freq_hz, Xl_mag, label="inductor |Xl|")
plt.loglog(freq_hz, Zrc_mag, label="series RC |Z|")

# Add readable labels and grid.
plt.title("Reactance and Impedance Magnitude Across Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (ohms)")
plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Show the comparison plot.
plt.legend()
plt.tight_layout()
plt.show()

```

## **3. Filter Response Basics**

### **3.1. Filter Response**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780964509" width="250">



>* Filters change signals differently by frequency
>* Response shows passed and reduced components

>* Capacitors and inductors react differently with frequency
>* Output location sets low-pass or high-pass behavior

>* Test filters with sine waves across frequencies
>* Response changes gradually in gain and phase



```
#@title Python Code - Filter Response

# Analyze a simple RC filter response.
# Complex numbers represent frequency dependent impedance.
# Gain phase and decibels describe behavior.

import math
import numpy as np
import matplotlib.pyplot as plt

# Define one resistor and capacitor.
R_ohms = 1000.0
C_farads = 1.0e-6
vin_volts = 1.0

# Choose frequencies around the cutoff.
frequencies_hz = np.logspace(1, 5, 400)
omega_rad_s = 2.0 * math.pi * frequencies_hz
z_capacitor = 1.0 / (1j * omega_rad_s * C_farads)

# Compute RC low pass transfer function.
h_lowpass = z_capacitor / (R_ohms + z_capacitor)
gain_linear = np.abs(h_lowpass)
phase_degrees = np.angle(h_lowpass, deg=True)

gain_db = 20.0 * np.log10(gain_linear)
cutoff_hz = 1.0 / (2.0 * math.pi * R_ohms * C_farads)
nearest_index = int(np.argmin(np.abs(frequencies_hz - cutoff_hz)))

# Show a few numerical checkpoints.
print("RC low-pass filter response example")
print(f"R = {R_ohms:.0f} ohms, C = {C_farads:.1e} F")
print(f"Cutoff frequency = {cutoff_hz:.1f} Hz")
print(f"Gain near cutoff = {gain_linear[nearest_index]:.3f}")

print(f"Gain near cutoff = {gain_db[nearest_index]:.2f} dB")
print(f"Phase near cutoff = {phase_degrees[nearest_index]:.1f} degrees")
print("Low frequencies pass, high frequencies are reduced.")

# Plot gain and phase versus frequency.
fig, axis_gain = plt.subplots(figsize=(8, 4.8))
axis_phase = axis_gain.twinx()
axis_gain.semilogx(frequencies_hz, gain_db, color="tab:blue")
axis_phase.semilogx(frequencies_hz, phase_degrees, color="tab:orange")

# Mark the cutoff for comparison.
axis_gain.axvline(cutoff_hz, color="gray", linestyle="--")
axis_gain.set_xlabel("Frequency (Hz)")
axis_gain.set_ylabel("Gain (dB)", color="tab:blue")
axis_phase.set_ylabel("Phase (degrees)", color="tab:orange")

axis_gain.set_title("RC Low-Pass Filter Response")
axis_gain.grid(True, which="both", linestyle=":")
fig.tight_layout()
plt.show()

```

### **3.2. Gain and Phase**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780964512" width="250">



>* Gain shows output amplitude versus input
>* Phase shows output timing shift

>* Filter gain varies with signal frequency.
>* Decibels describe attenuation or near-unity gain.

>* Phase affects timing and signal interaction
>* Transition regions change both gain and phase



```
#@title Python Code - Gain and Phase

# Demonstrate gain and phase for one filter.
# Complex impedances make AC ratios straightforward.
# Frequency changes both amplitude and timing.

import numpy as np
import matplotlib.pyplot as plt

# Define a simple RC low pass circuit.
resistance_ohms = 1000.0
capacitance_farads = 1.0e-6

# Choose logarithmically spaced test frequencies.
frequencies_hz = np.logspace(1, 5, 400)
omega_rad_s = 2.0 * np.pi * frequencies_hz

# Use complex capacitor impedance at every frequency.
capacitor_z = 1.0 / (1j * omega_rad_s * capacitance_farads)
transfer = capacitor_z / (resistance_ohms + capacitor_z)

# Convert complex response into gain and phase.
gain_ratio = np.abs(transfer)
gain_db = 20.0 * np.log10(gain_ratio)
phase_degrees = np.angle(transfer, deg=True)

# Mark the theoretical cutoff frequency.
cutoff_hz = 1.0 / (2.0 * np.pi * resistance_ohms * capacitance_farads)
index = int(np.argmin(np.abs(frequencies_hz - cutoff_hz)))

# Print a compact numerical summary.
print(f"RC low-pass cutoff frequency: {cutoff_hz:.1f} Hz")
print(f"Gain near cutoff: {gain_ratio[index]:.3f} ratio")
print(f"Gain near cutoff: {gain_db[index]:.2f} dB")
print(f"Phase near cutoff: {phase_degrees[index]:.1f} degrees")

# Plot gain and phase together.
fig, axis_gain = plt.subplots(figsize=(8, 4))
axis_gain.semilogx(frequencies_hz, gain_db, color="tab:blue")
axis_gain.set_xlabel("Frequency (Hz)")

# Label the gain axis.
axis_gain.set_ylabel("Gain (dB)", color="tab:blue")
axis_gain.tick_params(axis="y", labelcolor="tab:blue")
axis_gain.grid(True, which="both", linestyle="--", alpha=0.4)

# Add phase on a second vertical axis.
axis_phase = axis_gain.twinx()
axis_phase.semilogx(frequencies_hz, phase_degrees, color="tab:red")
axis_phase.set_ylabel("Phase (degrees)", color="tab:red")
axis_phase.tick_params(axis="y", labelcolor="tab:red")

# Highlight the cutoff transition region.
axis_gain.axvline(cutoff_hz, color="black", linestyle=":")
axis_gain.set_title("RC Low-Pass Filter Gain and Phase")
fig.tight_layout()
plt.show()

```

### **3.3. Phase Response**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780964510" width="250">



>* Filters shift signal timing by frequency
>* Phase affects waveform alignment and synchronization

>* Filter phase depends on circuit type
>* Phase shifts can affect real system timing

>* Gain, decibels, and phase describe filters
>* Phase shows timing changes near cutoff



```
#@title Python Code - Phase Response

# Phase response shows filter timing shifts.
# Complex numbers represent AC transfer functions.
# One plot compares gain and phase.

import numpy as np
import matplotlib.pyplot as plt

# Define a simple RC low-pass filter.
resistance_ohms = 1000.0
capacitance_farads = 1e-6
cutoff_hz = 1.0 / (2.0 * np.pi * resistance_ohms * capacitance_farads)

# Create a compact logarithmic frequency sweep.
frequencies_hz = np.logspace(1, 5, 400)
omega_rad_s = 2.0 * np.pi * frequencies_hz
complex_j = 1j

# Compute the RC low-pass transfer function.
transfer = 1.0 / (1.0 + complex_j * omega_rad_s * resistance_ohms * capacitance_farads)
gain = np.abs(transfer)
phase_degrees = np.angle(transfer, deg=True)

# Convert gain into decibels safely.
gain_db = 20.0 * np.log10(np.maximum(gain, 1e-12))

# Print a few representative phase values.
print(f"RC low-pass cutoff frequency: {cutoff_hz:.1f} Hz")
print(f"Phase near low frequency: {phase_degrees[0]:.1f} degrees")
print(f"Phase near cutoff: {-45.0:.1f} degrees approximately")
print(f"Phase near high frequency: {phase_degrees[-1]:.1f} degrees")

# Prepare one figure with gain and phase axes.
fig, gain_axis = plt.subplots(figsize=(8, 4.8))
phase_axis = gain_axis.twinx()

# Plot magnitude and phase together.
gain_axis.semilogx(frequencies_hz, gain_db, color="tab:blue", label="Gain")
phase_axis.semilogx(frequencies_hz, phase_degrees, color="tab:red", label="Phase")
gain_axis.axvline(cutoff_hz, color="gray", linestyle="--", linewidth=1.0)

# Label the shared frequency plot clearly.
gain_axis.set_xlabel("Frequency (Hz)")
gain_axis.set_ylabel("Gain (dB)", color="tab:blue")
phase_axis.set_ylabel("Phase shift (degrees)", color="tab:red")
gain_axis.set_title("RC Low-Pass Filter: Gain and Phase Response")

# Improve readability for beginner interpretation.
gain_axis.grid(True, which="both", linestyle=":", linewidth=0.7)
gain_axis.tick_params(axis="y", labelcolor="tab:blue")
phase_axis.tick_params(axis="y", labelcolor="tab:red")

# Show the figure in Google Colab.
fig.tight_layout()
plt.show()

```
