## **1. Safety Limit Checks**

### **1.1. Current Limit Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780959659" width="250">



>* Compare circuit current with safe ratings
>* Take action when limits are exceeded

>* Account for real-world current changes
>* Use conditions for safe overload responses

>* Test many circuit conditions automatically
>* Catch unsafe current before deployment



```
#@title Python Code - Current Limit Checks

# Check current against component ratings.
# Conditional statements protect electrical designs.
# Small examples keep safety logic clear.

# Define expected operating current values.
normal_current_a = 1.8
startup_current_a = 3.4
fault_current_a = 6.2

# Define the safe continuous current limit.
continuous_limit_a = 2.5
short_surge_limit_a = 4.0
print("Current limit check example")

# Check normal operating current safety.
if normal_current_a <= continuous_limit_a:
    normal_status = "SAFE for continuous operation"
else:
    normal_status = "UNSAFE for continuous operation"

# Check startup current as a short surge.
if startup_current_a <= short_surge_limit_a:
    startup_status = "acceptable short startup surge"
else:
    startup_status = "startup surge is too high"

# Check fault current against all limits.
if fault_current_a > short_surge_limit_a:
    fault_status = "shutdown required immediately"
elif fault_current_a > continuous_limit_a:
    fault_status = "reduce load or current limit"
else:
    fault_status = "current remains acceptable"

# Print concise engineering decisions.
print(f"Normal current: {normal_current_a} A -> {normal_status}")
print(f"Startup current: {startup_current_a} A -> {startup_status}")
print(f"Fault current: {fault_current_a} A -> {fault_status}")
print("Design action: choose ratings above worst credible current.")

```

### **1.2. Voltage Limit Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780959661" width="250">



>* Compare voltages with safe component limits
>* Trigger warnings or actions before overstress

>* Real voltages vary from ideal values
>* Conditionals separate normal, warning, and fault states

>* Voltage checks prevent unsafe circuit conditions
>* Conditional logic classifies design voltage risks



```
#@title Python Code - Voltage Limit Checks

# Conditional checks protect voltage rated components.
# This example classifies measured supply voltage.
# Safety margins help guide engineering decisions.

# Store measured voltages from a small test.
measured_voltages = [11.8, 12.3, 13.1, 14.6, 15.2]

# Define nominal and maximum safe voltage.
nominal_voltage = 12.0
maximum_voltage = 15.0

# Define warning threshold before the limit.
warning_voltage = 0.90 * maximum_voltage

# Check each voltage against design limits.
for sample_number, voltage in enumerate(measured_voltages, start=1):

    # Decide the safety status using conditionals.
    if voltage > maximum_voltage:
        status = "FAULT: shut down output"
    elif voltage >= warning_voltage:
        status = "WARNING: voltage near limit"
    else:
        status = "OK: voltage within range"

    # Compute percent margin below maximum rating.
    margin_percent = 100 * (maximum_voltage - voltage) / maximum_voltage

    # Print one concise result per sample.
    print(f"Sample {sample_number}: {voltage:.1f} V, {status}, margin {margin_percent:.1f}%")

# Summarize the design rule used.
print(f"Limit rule: safe at or below {maximum_voltage:.1f} V.")

```

### **1.3. Power Limit Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780959663" width="250">



>* Check power to prevent component overheating
>* Use conditions to trigger safe responses

>* Consider real operating conditions, not ratings alone
>* Use conditions to handle overload levels

>* Conditionals trigger protective power responses
>* Checks make designs safer and predictable



```
#@title Python Code - Power Limit Checks

# Check resistor power before accepting operation.
# Conditions compare calculated power with limits.
# Small tables keep the example readable.

# Define candidate operating points safely.
resistor_values_ohm = [100, 220, 470, 1000]
voltage_values_volt = [3.3, 5.0, 9.0, 12.0]
power_limit_watt = 0.25

# Print a short heading for learners.
print("Power limit check for quarter-watt resistors")
print("Ohms | Volts | Power W | Decision")

# Test each proposed resistor choice.
for resistance_ohm, voltage_volt in zip(resistor_values_ohm, voltage_values_volt):
    power_watt = voltage_volt ** 2 / resistance_ohm
    percent_rating = 100 * power_watt / power_limit_watt

    # Choose a safety message using conditions.
    if power_watt <= 0.70 * power_limit_watt:
        decision = "OK"
    elif power_watt <= power_limit_watt:
        decision = "Warning"
    else:
        decision = "Over limit"

    # Show one compact result line.
    result_line = f"{resistance_ohm:4d} | {voltage_volt:5.1f} | {power_watt:7.3f} | {decision}"
    print(result_line)

# Summarize the engineering rule applied.
safe_margin_percent = 70
summary_text = f"OK means below {safe_margin_percent}% of rated power."
print(summary_text)

```

## **2. Loop Patterns**

### **2.1. For Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780959651" width="250">



>* Repeat calculations across defined value sets
>* Organize inputs, operations, and results

>* Iterate through each sequence item
>* Repeat analyses across frequencies or samples

>* Test cases reveal trends and limits
>* Repeated analysis supports engineering design decisions



```
#@title Python Code - For Loops

# Demonstrate for loops with resistor choices.
# Repeat one circuit calculation systematically.
# Keep output short for Colab.

# Define a fixed supply voltage.
supply_voltage = 5.0

# Store candidate resistor values.
resistors_ohms = [100, 220, 470, 1000]

# Prepare a results container.
current_results = []

# Loop through each resistor value.
for resistance in resistors_ohms:
    current_amps = supply_voltage / resistance
    current_milliamps = current_amps * 1000
    current_results.append(current_milliamps)

# Print a compact table header.
print("Resistor sweep for a 5 V source")

# Display each repeated calculation result.
for resistance, current in zip(resistors_ohms, current_results):
    message = f"R = {resistance:4d} ohms -> I = {current:5.2f} mA"
    print(message)

# Summarize the looped engineering comparison.
best_current = min(current_results)
print(f"Lowest current tested: {best_current:.2f} mA")

```

### **2.2. While Loop Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780959649" width="250">



>* Repeat calculations until a condition changes
>* Useful for threshold-driven engineering simulations

>* Update loop variables each pass
>* Ensure conditions lead to completion

>* Stop when requirements or safety limits are met
>* Choose clear conditions, steps, and maximum limits



```
#@title Python Code - While Loop Patterns

# While loops repeat until conditions change.
# Electrical simulations often stop at thresholds.
# Safety guards prevent runaway loop execution.

# Define capacitor charging circuit constants.
supply_voltage = 5.0
resistance_ohms = 10000.0
capacitance_farads = 0.0001

# Define stopping threshold and time step.
threshold_voltage = 3.2
time_step_seconds = 0.05
max_steps_allowed = 500

# Start simulation state before looping.
time_seconds = 0.0
capacitor_voltage = 0.0
step_count = 0

# Continue while voltage remains below threshold.
while capacitor_voltage < threshold_voltage and step_count < max_steps_allowed:
    charging_rate = (supply_voltage - capacitor_voltage) / (resistance_ohms * capacitance_farads)
    capacitor_voltage = capacitor_voltage + charging_rate * time_step_seconds
    time_seconds = time_seconds + time_step_seconds

    step_count = step_count + 1

# Decide whether the loop stopped safely.
if capacitor_voltage >= threshold_voltage:
    stop_reason = "threshold voltage reached"
else:
    stop_reason = "maximum step guard reached"

# Print a compact engineering summary.
print("While loop capacitor charging example")
print(f"Stop reason: {stop_reason}")
print(f"Elapsed time: {time_seconds:.2f} s")
print(f"Final voltage: {capacitor_voltage:.3f} V")
print(f"Loop passes: {step_count}")

```

### **2.3. Resistor Iteration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780959652" width="250">



>* Test many standard resistor values systematically
>* Identify feasible, marginal, and rejected choices

>* Test candidate resistors in consistent calculations
>* Compare circuit results and reveal design trends

>* Consider real resistor limits and variations
>* Use loops to test practical design choices



```
#@title Python Code - Resistor Iteration

# Loop patterns compare practical resistor choices.
# Each candidate value gets tested consistently.
# Decisions mark safe and useful options.

# Define supply and LED design limits.
supply_voltage = 5.0
led_voltage = 2.0
minimum_current_ma = 5.0
maximum_current_ma = 20.0

# Use a small standard resistor list.
resistor_values_ohm = [150, 220, 330, 470, 680, 1000]
safe_choices = []
print("Resistor iteration for an LED limiter")
print("R(ohm)  I(mA)   decision")

# Repeat the same calculation for each resistor.
for resistor_ohm in resistor_values_ohm:
    current_amp = (supply_voltage - led_voltage) / resistor_ohm
    current_ma = current_amp * 1000.0
    within_limits = minimum_current_ma <= current_ma <= maximum_current_ma

    # Use a decision to label each result.
    if within_limits:
        decision = "acceptable"
        safe_choices.append(resistor_ohm)
    else:
        decision = "reject"

    # Print one compact line per candidate.
    result_line = f"{resistor_ohm:6d}  {current_ma:5.1f}   {decision}"
    print(result_line)

# Summarize the loop search outcome.
best_choice = safe_choices[0] if safe_choices else None
summary_text = f"First acceptable resistor: {best_choice} ohm"
print(summary_text)

```

## **3. Constraint Searches**

### **3.1. Threshold Search**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780959654" width="250">



>* Find where a design limit changes status
>* Loop through values and stop at threshold

>* Step through values in a planned direction
>* Stop when the constraint changes status

>* Choose step size and stopping rules carefully
>* Apply safety margins and engineering judgment



```
#@title Python Code - Threshold Search

# Threshold searches combine loops and decisions.
# This example checks capacitor ripple limits.
# The loop stops at first success.

# Define the load and ripple requirement.
load_current_amp = 0.08
frequency_hz = 120.0
max_ripple_volt = 0.10

# List practical candidate capacitance values.
candidates_microfarad = [100, 220, 330, 470, 680, 1000, 1500]
chosen_microfarad = None
chosen_ripple_volt = None

# Search until the constraint is satisfied.
for capacitance_microfarad in candidates_microfarad:
    capacitance_farads = capacitance_microfarad * 1e-6
    ripple_volt = load_current_amp / (frequency_hz * capacitance_farads)

    # Stop when ripple is acceptable.
    if ripple_volt <= max_ripple_volt:
        chosen_microfarad = capacitance_microfarad
        chosen_ripple_volt = ripple_volt
        break

# Report the engineering decision clearly.
print("Threshold search for capacitor ripple")
print(f"Ripple limit: {max_ripple_volt:.2f} V")
print(f"Load current: {load_current_amp:.2f} A")
print(f"Tested values: {candidates_microfarad}")

# Handle both success and failure safely.
if chosen_microfarad is None:
    print("No listed capacitor satisfies the limit.")
else:
    print(f"First acceptable value: {chosen_microfarad} microfarads")
    print(f"Estimated ripple: {chosen_ripple_volt:.3f} V")

```

### **3.2. Filter Feasible Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780959656" width="250">



>* Check each candidate against all constraints
>* Keep passing values, reject failing ones

>* Screen many candidates against required limits
>* Optimize only after finding workable choices

>* Make constraint checks consistent and repeatable
>* Support safer, clearer engineering design choices



```
#@title Python Code - Filter Feasible Values

# Search candidate resistors using repeated checks.
# Keep values satisfying every design limit.
# This mirrors practical electrical feasibility filtering.

# Define circuit and safety limits.
supply_voltage = 5.0
led_drop_voltage = 2.0
max_current_amp = 0.020

# Define acceptable operating current range.
min_current_amp = 0.010
max_power_watt = 0.125
standard_resistors_ohm = [100, 150, 180, 220, 270, 330, 390, 470]

# Store candidates that pass every condition.
feasible_resistors_ohm = []
for resistor_ohm in standard_resistors_ohm:
    voltage_across_resistor = supply_voltage - led_drop_voltage

    # Calculate current and resistor power.
    current_amp = voltage_across_resistor / resistor_ohm
    power_watt = current_amp ** 2 * resistor_ohm

    # Keep only values meeting all limits.
    current_is_safe = current_amp <= max_current_amp
    current_is_useful = current_amp >= min_current_amp
    power_is_safe = power_watt <= max_power_watt

    # Append candidates that pass every test.
    if current_is_safe and current_is_useful and power_is_safe:
        feasible_resistors_ohm.append((resistor_ohm, current_amp, power_watt))

# Print a compact feasibility summary.
print("Feasible LED resistor values:")
for resistor_ohm, current_amp, power_watt in feasible_resistors_ohm:
    milliamps = current_amp * 1000

    # Print one concise line per feasible value.
    print(f"{resistor_ohm} ohm: {milliamps:.1f} mA, {power_watt:.3f} W")

# Select a simple preferred feasible value.
best_choice = feasible_resistors_ohm[0] if feasible_resistors_ohm else None
print(f"First feasible choice: {best_choice[0]} ohm")

```

### **3.3. Valid Value Search**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780959657" width="250">



>* Loop through possible design values.
>* Use conditions to find acceptable choices.

>* Valid values must meet multiple constraints
>* Clear criteria define safe, useful choices

>* Find one valid option or optimize choices
>* Test real catalog values against constraints



```
#@title Python Code - Valid Value Search

# Search standard resistors for safe loading.
# Decisions test current and power limits.
# Loops examine each available candidate.

# No additional packages are required.

# Define available standard load resistors.
resistors_ohm = [100, 150, 220, 330, 470, 680, 1000]

# Define supply and engineering limits.
supply_volts, max_current_amp, max_power_watt = 5.0, 0.025, 0.125

# Store every resistor meeting all constraints.
valid_resistors = []

# Test each candidate with calculated quantities.
for resistor in resistors_ohm:
    current_amp = supply_volts / resistor
    power_watt = supply_volts * current_amp

    # Accept only values satisfying both limits.
    if current_amp <= max_current_amp and power_watt <= max_power_watt:
        valid_resistors.append((resistor, current_amp, power_watt))

# Choose the lowest valid resistance if available.
best_choice = valid_resistors[0] if valid_resistors else None

# Print a compact engineering search summary.
print("Valid value search for a 5 V load resistor")
print(f"Candidates tested: {len(resistors_ohm)} standard values")
print(f"Valid candidates found: {len(valid_resistors)} values")

# Report the first acceptable resistor clearly.
if best_choice is not None:
    resistor, current_amp, power_watt = best_choice
    print(f"First valid choice: {resistor} ohm")
    print(f"Current: {current_amp * 1000:.1f} mA")
    print(f"Power: {power_watt * 1000:.1f} mW")

# Provide a safe message when nothing passes.
else:
    print("No resistor satisfies all constraints.")

# Show all valid values without excessive output.
print("Valid resistor list:", [item[0] for item in valid_resistors])

```
