## **1. Function Basics**

### **1.1. Define Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780960577" width="250">



>* Functions package electrical calculations for reuse
>* Inputs produce consistent engineering results

>* Functions make scripts clearer and organized
>* Names and inputs support team reuse

>* Functions turn inputs into reusable results
>* Repeated calculations become clearer and easier to test



```
#@title Python Code - Define Functions

# Functions package repeated engineering calculations.
# Parameters carry values into functions.
# Return values send results back.

def power_from_current_resistance(current_amp, resistance_ohm):
    # Compute resistor power using I squared R.
    power_watt = current_amp**2 * resistance_ohm
    # Return the calculated engineering result.
    return power_watt

# Define a reusable voltage drop function.
def voltage_drop(current_amp, resistance_ohm):
    # Compute voltage drop using Ohm's law.
    drop_volt = current_amp * resistance_ohm
    # Return the voltage drop result.
    return drop_volt

# Store example circuit parameter values.
load_current_amp = 2.0
resistor_value_ohm = 8.0
wire_resistance_ohm = 0.35

# Call functions with electrical parameters.
resistor_power_watt = power_from_current_resistance(load_current_amp, resistor_value_ohm)
wire_drop_volt = voltage_drop(load_current_amp, wire_resistance_ohm)

# Print concise engineering results.
print("Function examples for electrical calculations")
print(f"Resistor power: {resistor_power_watt:.2f} W")
print(f"Wire voltage drop: {wire_drop_volt:.2f} V")
print("Defined functions can be reused safely.")

```

### **1.2. Current Parameter**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780960578" width="250">



>* Current input enables engineering calculations
>* Reusable functions handle many circuit cases

>* Current values depend on operating conditions
>* Explicit parameters reveal assumptions in calculations

>* Name current clearly and specify units
>* Reusable functions support many circuit calculations



```
#@title Python Code - Current Parameter

# Functions can accept current as input.
# Electrical calculations become reusable and clearer.
# Current units must be handled carefully.

# Define a function using current amperes.
def power_watts(voltage_volts, current_amps):
    return voltage_volts * current_amps

# Define another function for milliamps input.
def power_from_milliamps(voltage_volts, current_milliamps):
    current_amps = current_milliamps / 1000
    return power_watts(voltage_volts, current_amps)

# Calculate examples for two devices.
sensor_power = power_from_milliamps(5, 40)
motor_power = power_watts(24, 1.8)

# Print compact engineering results.
print("Sensor current: 40 mA at 5 V")
print("Sensor power:", round(sensor_power, 3), "W")

# Print a larger current example.
print("Motor current: 1.8 A at 24 V")
print("Motor power:", round(motor_power, 2), "W")

# Show why parameter units matter.
wrong_power = power_watts(5, 40)
print("If 40 mA is entered as 40 A:", wrong_power, "W")

```

### **1.3. Resistance Parameter**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780960580" width="250">



>* Resistance inputs enable electrical calculations
>* Reusable functions handle changing resistance values

>* Define resistance with clear units and meaning
>* Reuse functions with different resistance inputs

>* Resistance varies with real operating conditions
>* Choose suitable values for dependable calculations



```
#@title Python Code - Resistance Parameter

# Functions make electrical calculations reusable.
# Resistance parameters represent circuit opposition.
# Return values keep results available.

# Define a function using resistance and current.
def voltage_drop(current_ampere, resistance_ohm):
    voltage_volt = current_ampere * resistance_ohm
    return voltage_volt

# Define another function using default current.
def power_loss(resistance_ohm, current_ampere=2.0):
    power_watt = current_ampere ** 2 * resistance_ohm
    return power_watt

# Store measured and catalog resistance values.
catalog_resistance_ohm = 10.0
measured_resistance_ohm = 10.8
load_current_ampere = 1.5

# Call functions with different resistance parameters.
catalog_voltage = voltage_drop(load_current_ampere, catalog_resistance_ohm)
measured_voltage = voltage_drop(load_current_ampere, measured_resistance_ohm)
default_power = power_loss(measured_resistance_ohm)

# Display concise engineering calculation results.
print("Resistance parameter function example")
print(f"Catalog voltage drop: {catalog_voltage:.2f} V")
print(f"Measured voltage drop: {measured_voltage:.2f} V")
print(f"Default-current power loss: {default_power:.2f} W")

```

## **2. Reusable Calculation Functions**

### **2.1. Voltage Parameter**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780960573" width="250">



>* Voltage inputs make functions reusable
>* One calculation supports many circuit scenarios

>* Voltage input defines the operating condition
>* Supports clear comparisons across circuit cases

>* Name voltage inputs by physical meaning
>* Choose the relevant voltage for reliable results



```
#@title Python Code - Voltage Parameter

# Voltage parameters make functions reusable.
# Electrical calculations should expose assumptions.
# Default arguments support flexible comparisons.

# Define a reusable current calculation.
def calculate_current(voltage_volts, resistance_ohms=1000):
    if resistance_ohms <= 0:
        raise ValueError("Resistance must be positive.")

    current_amps = voltage_volts / resistance_ohms
    return current_amps

# Compare different voltage operating points.
bench_voltage = 5.0
battery_voltage = 3.7
adapter_voltage = 12.0

# Use the same resistance default each time.
bench_current = calculate_current(bench_voltage)
battery_current = calculate_current(battery_voltage)
adapter_current = calculate_current(adapter_voltage)

# Print concise engineering results.
print("Reusable voltage parameter example")
print(f"5.0 V source current: {bench_current:.4f} A")
print(f"3.7 V battery current: {battery_current:.4f} A")
print(f"12.0 V adapter current: {adapter_current:.4f} A")

# Override the default resistance parameter.
sensor_current = calculate_current(3.3, resistance_ohms=2200)
print(f"3.3 V sensor current: {sensor_current:.4f} A")

```

### **2.2. Default Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780960571" width="250">



>* Defaults supply common engineering assumptions
>* Functions stay simple yet adaptable

>* Defaults show expected engineering assumptions
>* Callers can override them for special conditions

>* Defaults simplify function calls and testing
>* Functions stay flexible for advanced scenarios



```
#@title Python Code - Default Parameters

# Default parameters simplify repeated engineering calculations.
# This example estimates resistor power ratings.
# Callers may override normal design assumptions.

# Define a reusable calculation function.
def required_resistor_power(voltage, resistance, safety_factor=2.0):
    actual_power = voltage**2 / resistance
    rated_power = actual_power * safety_factor

    return actual_power, rated_power

# Use the default safety factor.
actual_normal, rating_normal = required_resistor_power(12, 100)

# Override the default for harsher operation.
actual_harsh, rating_harsh = required_resistor_power(12, 100, 3.0)

# Show both calculations clearly.
print("Default safety factor calculation:")
print(f"Actual power: {actual_normal:.2f} W")
print(f"Recommended rating: {rating_normal:.2f} W")

# Compare against the overridden default.
print("Harsh environment calculation:")
print(f"Actual power: {actual_harsh:.2f} W")
print(f"Recommended rating: {rating_harsh:.2f} W")

```

### **2.3. Resistance Parameter**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780960575" width="250">



>* Resistance input models current opposition
>* One function handles many component cases

>* Pass resistance as meaningful calculation input
>* Reuse functions with resistance from any source

>* Defaults speed common resistance calculations
>* Document assumptions to avoid misleading results



```
#@title Python Code - Resistance Parameter

# Functions can reuse electrical calculation logic.
# Resistance parameters make formulas more flexible.
# Default arguments support common engineering assumptions.

# Define current using Ohm's law.
def current_from_voltage(voltage_volts, resistance_ohms=1000):
    if resistance_ohms <= 0:
        raise ValueError("Resistance must be positive.")

    current_amps = voltage_volts / resistance_ohms
    return current_amps

# Define voltage drop across wiring resistance.
def voltage_drop(current_amps, resistance_ohms=0.25):
    if resistance_ohms < 0:
        raise ValueError("Resistance cannot be negative.")

    drop_volts = current_amps * resistance_ohms
    return drop_volts

# Define a few component design cases.
cases = [
    ("LED test load", 5.0, 1000),
    ("Heater element", 120.0, 24),
    ("Sensor input", 3.3, 2200),
]

# Print reusable function results.
print("Resistance parameter examples")
for name, voltage, resistance in cases:
    amps = current_from_voltage(voltage, resistance)

    milliamps = amps * 1000
    print(f"{name}: {milliamps:.2f} mA")

# Use default resistance for a common test load.
default_amps = current_from_voltage(5.0)
print(f"Default 1000 ohm load: {default_amps * 1000:.2f} mA")

# Compare wiring drops using metric-style lengths.
short_wire_drop = voltage_drop(2.0, resistance_ohms=0.08)
long_wire_drop = voltage_drop(2.0, resistance_ohms=0.40)

# Print voltage drop comparison results.
print(f"Short cable drop: {short_wire_drop:.2f} V")
print(f"Long cable drop: {long_wire_drop:.2f} V")

```

## **3. Simple Python Modules**

### **3.1. Module Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780960566" width="250">



>* Modules store reusable Python definitions
>* Shared calculations stay organized and trusted

>* Focused modules organize related engineering tools
>* Reusable functions keep scripts clean and maintainable

>* Clear module functions reduce engineering mistakes
>* Reusable modules support dependable shared analysis



```
#@title Python Code - Module Basics

# Modules store reusable engineering calculations.
# This example creates one tiny module.
# Then another script imports its tools.

# Import tools for creating a temporary module.
import importlib.util as module_loader

# Import temporary file utilities from Python.
import tempfile as temporary_files

# Create readable module code as text.
module_text = """
OHM_SYMBOL = 'ohm'


def current_amps(voltage_volts, resistance_ohms):
    return voltage_volts / resistance_ohms


def power_watts(voltage_volts, current_amps):
    return voltage_volts * current_amps


def series_resistance(*resistors_ohms):
    return sum(resistors_ohms)
"""

# Store the module in a temporary file.
module_file = temporary_files.NamedTemporaryFile(
    mode="w",
    suffix=".py",
    delete=False,
)

# Write module code, then close file.
module_file.write(module_text)
module_file.close()

# Build an import specification for the module.
specification = module_loader.spec_from_file_location(
    "circuit_tools",
    module_file.name,
)

# Create a module object from specification.
circuit_tools = module_loader.module_from_spec(specification)
specification.loader.exec_module(circuit_tools)

# Use imported functions in the main script.
voltage_volts = 12.0
resistance_ohms = circuit_tools.series_resistance(100, 220, 330)
current_value = circuit_tools.current_amps(voltage_volts, resistance_ohms)

# Calculate power using another module function.
power_value = circuit_tools.power_watts(voltage_volts, current_value)
module_name = circuit_tools.__name__

# Print a compact engineering summary.
print(f"Imported module name: {module_name}")
print(f"Series resistance: {resistance_ohms:.1f} ohms")
print(f"Circuit current: {current_value:.4f} A")
print(f"Circuit power: {power_value:.4f} W")

```

### **3.2. Grouping Reusable Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780960568" width="250">



>* Collect common calculations in shared modules
>* Reuse trusted functions across engineering scripts

>* Group functions by shared engineering purpose
>* Balanced modules improve readability and reuse

>* Modules support shared engineering workflows
>* Updates improve all dependent scripts



```
#@title Python Code - Grouping Reusable Functions

# Grouped functions can form reusable engineering modules.
# This example writes one tiny module file.
# Another script imports and uses those functions.

# Import standard tools for module file handling.
from pathlib import Path

# Store module code as a reusable text block.
module_code = '''
def resistance_from_voltage_current(voltage, current):
    if current == 0:
        raise ValueError("current must not be zero")
    return voltage / current

def dc_power(voltage, current):
    return voltage * current

def voltage_drop(current, resistance, length_m=1.0):
    return current * resistance * length_m
'''

# Write the small module beside this notebook session.
Path("circuit_tools.py").write_text(module_code)

# Import related functions from the grouped module.
from circuit_tools import dc_power, resistance_from_voltage_current, voltage_drop

# Define a small circuit example using metric length.
voltage_v, current_a, cable_m = 12.0, 2.0, 3.0

# Call functions as reusable engineering tools.
load_r = resistance_from_voltage_current(voltage_v, current_a)
load_p = dc_power(voltage_v, current_a)
drop_v = voltage_drop(current_a, 0.08, length_m=cable_m)

# Print a compact summary of reusable results.
print("Reusable module: circuit_tools.py")
print(f"Load resistance: {load_r:.2f} ohms")
print(f"Load power: {load_p:.2f} watts")
print(f"Cable voltage drop: {drop_v:.2f} volts")

```

### **3.3. Testing Module Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780960569" width="250">



>* Test functions before reusing engineering modules
>* Build confidence in reliable calculation tools

>* Start with simple, known test cases
>* Check edge cases and unusual inputs

>* Tests protect modules during future changes
>* Reliable functions support safe engineering decisions



```
#@title Python Code - Testing Module Functions

# Testing confirms reusable electrical module functions.
# Simple checks compare outputs to expectations.
# Edge cases reveal hidden calculation problems.

# Import tools for temporary module creation.
import importlib.util, pathlib, tempfile

# Create a temporary module file.
module_text = """
def dc_power(voltage, current):
    return voltage * current

def series_resistance(*resistors):
    return sum(resistors)

def voltage_drop(current, resistance, lead_length_m=1.0):
    return current * resistance * lead_length_m
"""

# Store module code in a safe location.
folder_path = pathlib.Path(tempfile.mkdtemp())
module_path = folder_path / "circuit_tools.py"
module_path.write_text(module_text)

# Load the module for reuse.
spec = importlib.util.spec_from_file_location(
    "circuit_tools", module_path
)
circuit_tools = importlib.util.module_from_spec(spec)
spec.loader.exec_module(circuit_tools)

# Define tiny expected engineering cases.
tests = [
    ("power", circuit_tools.dc_power(12, 2), 24),
    ("series", circuit_tools.series_resistance(100, 220), 320),
    ("drop", circuit_tools.voltage_drop(2, 0.05, 10), 1.0),
    ("zero current", circuit_tools.dc_power(5, 0), 0),
]

# Compare results with known values.
passed = 0
for name, actual, expected in tests:
    ok = abs(actual - expected) < 1e-9
    passed += int(ok)
    print(f"{name}: expected {expected}, got {actual}, pass={ok}")

# Summarize testing before reuse.
print(f"Passed {passed} of {len(tests)} module tests.")

```
