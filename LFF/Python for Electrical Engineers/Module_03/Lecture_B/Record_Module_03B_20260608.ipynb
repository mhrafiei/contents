## **1. Reading Measurement Data**

### **1.1. Opening Data Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1780962518" width="250">



>* Open files to access stored measurements
>* Choose readable sources and correct access modes

>* Check file path, name, and purpose
>* Open measurements read-only to protect data

>* Plan for missing or inaccessible files
>* Close files for reliable measurement analysis



```
#@title Python Code - Opening Data Files

# Measurement files connect stored values to analysis.
# This example opens a small voltage file.
# Built in handling keeps reading controlled.

# Create a tiny measurement file.
file_name = "voltage_readings.txt"

# Prepare sample multimeter readings.
readings = ["1.98", "2.01", "2.00"]

# Write readings for this Colab demonstration.
with open(file_name, "w", encoding="utf-8") as data_file:
    data_file.write("\n".join(readings))

# Open the file safely for reading.
try:
    with open(file_name, "r", encoding="utf-8") as data_file:
        text_lines = data_file.readlines()

# Report missing file problems clearly.
except FileNotFoundError:
    text_lines = []
    print("Measurement file was not found.")

# Convert text measurements into numbers.
voltages = [float(line.strip()) for line in text_lines]

# Summarize the opened measurement file.
if voltages:
    average_voltage = sum(voltages) / len(voltages)
    print(f"Opened file: {file_name}")
    print(f"Readings loaded: {len(voltages)} values")
    print(f"Average voltage: {average_voltage:.3f} V")

# Show the first measurement value.
print(f"First reading: {voltages[0]:.2f} V" if voltages else "No readings loaded.")

```

### **1.2. Loading Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1780962520" width="250">



>* Load text measurements for analysis and reporting
>* Identify numbers, labels, units, blanks, comments

>* Read each measurement record line by line
>* Split separators and convert text to numbers

>* Expect messy measurement files and extra information
>* Load clean values for trusted analysis



```
#@title Python Code - Loading Measurements

# Load simple measurement files safely.
# Convert text readings into numbers.
# Summarize engineering data with units.

# Create a small measurement text file.
file_name = "voltage_measurements.txt"

# Store comments, blanks, and readings.
file_text = """# DMM voltage log, units volts
# resistor divider test
1.195
1.201

bad_sample
1.198
1.204
"""

# Write the example file locally.
with open(file_name, "w") as file_object:
    file_object.write(file_text)

# Prepare containers for useful results.
voltages = []
skipped_lines = []

# Read the measurement file line by line.
with open(file_name, "r") as file_object:
    for line_number, raw_line in enumerate(file_object, start=1):
        text = raw_line.strip()

        # Ignore blank lines and comments.
        if text == "" or text.startswith("#"):
            continue

        # Try converting text into numeric data.
        try:
            voltage_value = float(text)
            voltages.append(voltage_value)
        except ValueError:
            skipped_lines.append(line_number)

# Validate that measurements were loaded.
if len(voltages) == 0:
    raise ValueError("No numeric measurements were loaded.")

# Calculate simple engineering summary values.
count = len(voltages)
average_voltage = sum(voltages) / count
minimum_voltage = min(voltages)
maximum_voltage = max(voltages)

# Print a compact loading report.
print("Loading Measurements Report")
print(f"Loaded samples: {count}")
print(f"Skipped data lines: {skipped_lines}")
print(f"Average voltage: {average_voltage:.3f} V")
print(f"Minimum voltage: {minimum_voltage:.3f} V")
print(f"Maximum voltage: {maximum_voltage:.3f} V")

```

### **1.3. File Reading Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1780962522" width="250">



>* Text files store raw measurement characters
>* Programs read, interpret, then convert values

>* Follow a careful file reading sequence
>* Separate data values from messy file text

>* Check files for location and format errors
>* Validate measurements before using them



```
#@title Python Code - File Reading Basics

# Learn file reading with measurement text.
# Convert raw strings into engineering numbers.
# Check loaded values before using them.

# Create a tiny measurement file safely.
file_name = "voltage_log.txt"

# Store one header and three readings.
file_text = "Voltage_V\n1.20\n1.25\n1.18\n"

# Write sample data using built-in handling.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write(file_text)

# Read the saved file as text.
with open(file_name, "r", encoding="utf-8") as file_object:
    raw_lines = file_object.readlines()

# Ignore the header line safely.
data_lines = raw_lines[1:]

# Convert clean text lines into floats.
voltages = [float(line.strip()) for line in data_lines]

# Validate that measurements were found.
if len(voltages) == 0:
    raise ValueError("No voltage readings were loaded")

# Compute simple engineering summary values.
average_voltage = sum(voltages) / len(voltages)
maximum_voltage = max(voltages)
minimum_voltage = min(voltages)

# Print compact labeled results with units.
print(f"Read {len(voltages)} voltage measurements.")
print(f"Average voltage: {average_voltage:.2f} V")
print(f"Minimum voltage: {minimum_voltage:.2f} V")
print(f"Maximum voltage: {maximum_voltage:.2f} V")
print("File reading complete and values verified.")

```

## **2. Writing Results**

### **2.1. Writing Text Reports**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1780962515" width="250">



>* Turn calculations into shareable records
>* Save portable reports for later review

>* Design reports for clear human understanding
>* Label values to prevent engineering mistakes

>* Organize reports with clear beginning, middle, end
>* Add context so results are understandable



```
#@title Python Code - Writing Text Reports

# This script writes a small engineering report.
# It formats circuit values with readable units.
# It then previews the saved text file.

# Define simple circuit input values.
source_voltage_v = 12.0
resistance_ohm = 470.0
component_rating_w = 0.25

# Calculate current and resistor power.
current_a = source_voltage_v / resistance_ohm
power_w = source_voltage_v * current_a
power_mw = power_w * 1000.0

# Choose a clear report filename.
report_name = "circuit_report.txt"
report_lines = []

# Build report sections as readable strings.
report_lines.append("Circuit Calculation Report")
report_lines.append("--------------------------")
report_lines.append(f"Source voltage: {source_voltage_v:.2f} V")

# Add labeled calculated engineering results.
report_lines.append(f"Resistance: {resistance_ohm:.1f} ohm")
report_lines.append(f"Current: {current_a * 1000:.2f} mA")
report_lines.append(f"Power: {power_mw:.2f} mW")

# Add a short engineering interpretation.
if power_w <= component_rating_w:
    status_text = "within the component rating"
else:
    status_text = "above the component rating"

# Finish the report with context.
report_lines.append(f"Rating check: {status_text}.")
report_lines.append("End of report.")
report_text = "\n".join(report_lines)

# Write the report using built-in file handling.
with open(report_name, "w", encoding="utf-8") as file_object:
    written_count = file_object.write(report_text)

# Read back the report for confirmation.
with open(report_name, "r", encoding="utf-8") as file_object:
    preview_text = file_object.read()

# Display a short confirmation and preview.
print(f"Saved {written_count} characters to {report_name}.")
print(preview_text)

```

### **2.2. Structured Text Reports**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1780962513" width="250">



>* Organize calculations into predictable report sections
>* Preserve context for review and troubleshooting

>* Consistent fields support comparison and archiving
>* Clear labels make reports dependable communication

>* Separate raw data from interpreted results
>* Guide readers from context to conclusion



```
#@title Python Code - Structured Text Reports

# Structured reports preserve engineering calculation context.
# Labels and units prevent misread results.
# This example writes one clear report.

# Define small measured voltage data.
time_s = [0.00, 0.25, 0.50, 0.75]
voltage_v = [0.10, 1.20, 2.40, 1.10]
resistance_ohm = 220.0

# Validate matching measurement list lengths.
if len(time_s) != len(voltage_v):
    raise ValueError("Measurement lists must match.")

# Calculate simple circuit and signal quantities.
peak_v = max(voltage_v)
average_v = sum(voltage_v) / len(voltage_v)
peak_current_a = peak_v / resistance_ohm

# Calculate power at the peak voltage.
peak_power_w = (peak_v ** 2) / resistance_ohm
limit_v = 3.30
status = "PASS" if peak_v <= limit_v else "FAIL"

# Store report lines in predictable sections.
report_lines = [
    "Voltage Signal Test Report",
    "==========================",
    "Test condition: resistor load signal check",
    f"Load resistance: {resistance_ohm:.1f} ohm",
]

# Add measurement summary lines with units.
report_lines += [
    "",
    "Input Measurements",
    f"Sample count: {len(voltage_v)} samples",
    f"Time range: {time_s[0]:.2f} to {time_s[-1]:.2f} s",
]

# Add calculated engineering results.
report_lines += [
    "",
    "Calculated Results",
    f"Peak voltage: {peak_v:.2f} V",
    f"Average voltage: {average_v:.2f} V",
]

# Add circuit outputs and conclusion.
report_lines += [
    f"Peak current: {peak_current_a * 1000:.2f} mA",
    f"Peak power: {peak_power_w * 1000:.2f} mW",
    f"Voltage limit check: {status}",
]

# Write the structured report to disk.
report_name = "structured_signal_report.txt"
with open(report_name, "w", encoding="utf-8") as report_file:
    report_file.write("\n".join(report_lines))

# Read back selected lines for display.
with open(report_name, "r", encoding="utf-8") as report_file:
    preview_lines = report_file.readlines()[:12]

# Print a compact preview of the report.
print("Structured report preview:")
print("".join(preview_lines))

```

### **2.3. Formatting Report Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1780962517" width="250">



>* Add labels, units, and context
>* Make results clear, useful, and traceable

>* Choose precision that matches measurement confidence
>* Use consistent rounding for easy comparisons

>* Include clear labels and units
>* Use readable prefixes to prevent mistakes



```
#@title Python Code - Formatting Report Values

# Format calculated engineering values clearly.
# Labels and units prevent ambiguity.
# Precision communicates realistic measurement confidence.

# Store calculated circuit and signal values.
cutoff_hz = 1247.83
power_w = 0.18364
vpp_v = 3.297
rise_time_s = 0.00000472

# Convert values into readable engineering units.
cutoff_khz = cutoff_hz / 1000
power_mw = power_w * 1000
rise_time_us = rise_time_s * 1000000

# Build formatted report lines consistently.
report_lines = [
    "Electrical Test Report",
    "Cutoff frequency : {0:7.2f} kHz".format(cutoff_khz),
    "Resistor power   : {0:7.1f} mW".format(power_mw),
]

# Add voltage and timing results.
report_lines.extend([
    "Waveform Vpp     : {0:7.2f} V".format(vpp_v),
    "Rise time        : {0:7.2f} us".format(rise_time_us),
    "Status           : values rounded for reporting",
])

# Join lines into one report string.
report_text = "\n".join(report_lines)

# Print the formatted report.
print(report_text)

# Save the same report to disk.
report_file = "formatted_results_report.txt"

# Write a small structured text report.
with open(report_file, "w") as file_object:
    file_object.write(report_text)

# Confirm where the report was saved.
print("Saved report file:", report_file)

```

## **3. Formatted Engineering Reports**

### **3.1. Frequency Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1780962508" width="250">



>* Format frequencies to show scale clearly
>* Use units and precision to prevent errors

>* Balance precision with readability
>* Match digits to measurement purpose

>* Use consistent frequency units within each report
>* Choose prefixes suited to the application



```
#@title Python Code - Frequency Formatting

# Format frequencies for readable engineering reports.
# Use prefixes, precision, and unit labels.
# Keep output short for lecture clarity.

# Store sample frequency measurements in hertz.
frequency_hz_values = [12.5, 1247.893, 1_000_000, 2_450_000_000]

# Match each range with a readable prefix.
scales = [(1e9, "GHz"), (1e6, "MHz"), (1e3, "kHz"), (1.0, "Hz")]

# Convert hertz into the largest suitable unit.
def format_frequency(value_hz):
    for factor, unit_label in scales:
        if abs(value_hz) >= factor:
            scaled_value = value_hz / factor
            return f"{scaled_value:8.2f} {unit_label}"

    return f"{value_hz:8.2f} Hz"

# Build formatted report lines with labels.
report_lines = ["Formatted Engineering Report", "Frequency Results"]

# Add each measurement using aligned columns.
for index_value, frequency_hz in enumerate(frequency_hz_values, start=1):
    formatted_text = format_frequency(frequency_hz)
    report_lines.append(f"Point {index_value}: {formatted_text}")

# Join the report into printable text.
report_text = "\n".join(report_lines)

# Display the compact engineering report.
print(report_text)

```

### **3.2. Units in Reports**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1780962510" width="250">



>* Units give numbers clear engineering meaning
>* Formatted reports identify each measured result

>* Clear units prevent scale mistakes
>* Consistent units support reliable engineering decisions

>* Choose units matching scale and audience
>* Keep unit style consistent throughout reports



```
#@title Python Code - Units in Reports

# Format engineering values with clear units.
# Reports should pair numbers and labels.
# Unit choices make results easier.

# Store calculated values in base units.
voltage_v = 5.0
current_a = 0.002
resistance_ohm = voltage_v / current_a

# Convert values into readable engineering units.
current_ma = current_a * 1000
resistance_kohm = resistance_ohm / 1000
power_mw = voltage_v * current_a * 1000

# Build report lines using formatted strings.
report_lines = []
report_lines.append("Formatted Engineering Report")
report_lines.append("-----------------------------")

# Include labels, precision, and units together.
report_lines.append(f"Voltage:    {voltage_v:6.2f} V")
report_lines.append(f"Current:    {current_ma:6.2f} mA")
report_lines.append(f"Resistance: {resistance_kohm:6.2f} kΩ")
report_lines.append(f"Power:      {power_mw:6.2f} mW")

# Add one signal example using hertz.
sampling_hz = 44100
sampling_khz = sampling_hz / 1000
report_lines.append(f"Sampling:   {sampling_khz:6.1f} kHz")

# Save the short report as text.
report_text = "\n".join(report_lines)
with open("engineering_report.txt", "w", encoding="utf-8") as file:
    file.write(report_text)

# Read back the report for inspection.
with open("engineering_report.txt", "r", encoding="utf-8") as file:
    saved_report = file.read()

# Display the formatted report.
print(saved_report)

```

### **3.3. Aligned Units**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1780962512" width="250">



>* Alignment makes engineering values easier to scan
>* Consistent units reduce report misreading

>* Align labels, values, and units consistently
>* Clear formatting supports accurate technical decisions

>* Formatting communicates engineering data clearly
>* Alignment improves comparison, auditing, and reliability



```
#@title Python Code - Aligned Units

# Format reports with aligned engineering units.
# Labels, values, and units share columns.
# This example uses only built-in Python.

# Store small circuit and signal results.
measurements = [
    ("Input voltage", 3.300, "V"),
    ("Sensor current", 0.0125, "A"),

    ("Load resistance", 220.0, "ohm"),
    ("Cutoff frequency", 1450.0, "Hz"),
    ("Cable length", 36.0, "in")
]

# Build one report line for each measurement.
report_lines = []
for label, value, unit in measurements:
    formatted = f"{label:<18} {value:>10.3f} {unit:<4}"
    report_lines.append(formatted)

# Save the aligned report to a text file.
report_text = "Formatted Engineering Report\n" + "\n".join(report_lines)
with open("aligned_units_report.txt", "w", encoding="utf-8") as file:
    file.write(report_text)

# Print the compact report for inspection.
print(report_text)

```
