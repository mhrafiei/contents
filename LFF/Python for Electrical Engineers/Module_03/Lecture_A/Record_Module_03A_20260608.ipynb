## **1. Lists and Tuples**

### **1.1. Voltage Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780961528" width="250">



>* Voltage readings form ordered time-based samples
>* Lists preserve changing system behavior

>* Ordered samples preserve voltage changes over time
>* Lists change; tuples keep completed data fixed

>* Lists show voltage data still changing
>* Tuples preserve ordered voltage records



```
#@title Python Code - Voltage Samples

# Lists store voltage samples in order.
# Tuples protect completed waveform sample records.
# Indexing and slicing summarize signal behavior.

# Store active measurements in a changeable list.
voltage_samples = [0.12, 1.18, 2.35, 3.44, 4.08]

# Append a new sample from continued logging.
voltage_samples.append(4.62)

# Correct one reading after probe recalibration.
voltage_samples[1] = 1.22

# Freeze completed readings as a tuple.
final_trace = tuple(voltage_samples)

# Extract important ordered sample information.
first_sample = final_trace[0]
last_sample = final_trace[-1]
middle_samples = final_trace[2:5]

# Aggregate values for engineering summary.
average_voltage = sum(final_trace) / len(final_trace)
peak_voltage = max(final_trace)
voltage_rise = last_sample - first_sample

# Print compact results for learners.
print("Active list:", voltage_samples)
print("Fixed tuple:", final_trace)
print("First and last:", first_sample, last_sample)
print("Middle slice:", middle_samples)
print("Average voltage:", round(average_voltage, 2), "V")
print("Peak voltage:", peak_voltage, "V")
print("Total rise:", round(voltage_rise, 2), "V")

```

### **1.2. Fixed Waveform Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780961529" width="250">



>* Ordered samples preserve waveform timing and shape
>* Use lists for editing, tuples for references

>* Fixed samples support repeatable waveform comparisons
>* Ordered values preserve signal behavior over time

>* Use lists while waveform data changes
>* Use tuples for reviewed reference waveforms



```
#@title Python Code - Fixed Waveform Samples

# Store waveform samples in ordered collections.
# Compare mutable lists with stable tuples.
# Summarize fixed samples using simple indexing.

# Define editable oscilloscope voltage samples.
samples_list = [0.0, 1.2, 2.0, 1.2, 0.0, -1.2, -2.0, -1.2]

# Correct a small probe offset.
offset_volts = 0.05

# Build corrected samples with a list.
corrected_list = [round(value - offset_volts, 2) for value in samples_list]

# Freeze the accepted reference waveform.
reference_tuple = tuple(corrected_list)

# Validate the waveform sample count.
expected_count = 8

# Check count before making summaries.
if len(reference_tuple) == expected_count:
    status_message = "sample count matches expected one-cycle record"

else:
    status_message = "sample count does not match expected record"

# Select important ordered positions.
first_sample = reference_tuple[0]
peak_sample = max(reference_tuple)
minimum_sample = min(reference_tuple)

# Slice the positive half-cycle samples.
positive_half = reference_tuple[0:4]

# Aggregate values for engineering summaries.
average_voltage = sum(reference_tuple) / len(reference_tuple)
peak_to_peak = peak_sample - minimum_sample

# Print a compact teaching summary.
print("Original list:", samples_list)
print("Fixed tuple:", reference_tuple)
print("Status:", status_message)
print("First sample V:", first_sample)
print("Positive half-cycle:", positive_half)
print("Average V:", round(average_voltage, 3))
print("Peak-to-peak V:", round(peak_to_peak, 3))

```

### **1.3. Current Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780961531" width="250">



>* Store current readings in ordered lists
>* Use order to reveal circuit behavior

>* Use tuples for fixed current samples
>* Protect approved data for repeatable analysis

>* Connect current order to measurement context
>* Summarize patterns using lists or tuples



```
#@title Python Code - Current Samples

# Store current samples in ordered collections.
# Lists can grow during laboratory measurements.
# Tuples protect approved reference sample sets.

# Create a list of measured motor currents.
current_samples_amps = [0.4, 1.8, 3.2, 2.1, 1.2, 0.9]

# Create a tuple of approved baseline currents.
baseline_amps = (0.5, 1.7, 3.0, 2.0, 1.1, 0.8)

# Validate matching sample counts before comparison.
same_length = len(current_samples_amps) == len(baseline_amps)

# Add a new reading while testing continues.
current_samples_amps.append(0.85)

# Select early transient samples using slicing.
early_samples = current_samples_amps[:3]

# Compute simple engineering summaries.
peak_current = max(current_samples_amps)
average_current = sum(current_samples_amps) / len(current_samples_amps)

# Compare measured and baseline peaks safely.
baseline_peak = max(baseline_amps) if same_length else None

# Print compact results for the lecture.
print("Measured current samples, A:", current_samples_amps)
print("Approved baseline tuple, A:", baseline_amps)
print("Early transient samples, A:", early_samples)
print("Peak measured current, A:", peak_current)
print("Average measured current, A:", round(average_current, 3))
print("Matching original counts:", same_length)
print("Baseline peak current, A:", baseline_peak)

```

## **2. Engineering Dictionaries**

### **2.1. Component Value Maps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780961533" width="250">



>* Map component labels to their values
>* Connect schematics, data, and real parts

>* Dictionaries keep component labels meaningful
>* Lists lose meaning without remembered positions

>* Verify components across reviews, troubleshooting, and testing
>* Preserve readable links between labels and values



```
#@title Python Code - Component Value Maps

# Dictionaries map labels to engineering values.
# Component labels preserve schematic design meaning.
# Small maps are easy to inspect.

# Create a component value dictionary.
component_values = {
    "R1": "10 kOhm",
    "R2": "2.2 kOhm",

    "C1": "100 nF",
    "C2": "10 uF",
    "L1": "47 uH",

    "D1": "1N4148",
}

# Read one value by reference designator.
selected_label = "C1"
selected_value = component_values[selected_label]

# Update a value after design review.
component_values["R2"] = "2.0 kOhm"

# Add a new labeled component.
component_values["U1"] = "LM358 op amp"

# Count parts using dictionary length.
part_count = len(component_values)

# Group labels by first character.
resistor_labels = [key for key in component_values if key.startswith("R")]
capacitor_labels = [key for key in component_values if key.startswith("C")]

# Print compact engineering summary lines.
print("Component value map example")
print(f"{selected_label} is {selected_value}")
print(f"Updated R2 is {component_values['R2']}")
print(f"Total labeled parts: {part_count}")

# Iterate through labels in sorted order.
for label in sorted(component_values):
    print(f"{label}: {component_values[label]}")

# Show filtered design groups.
print(f"Resistor labels: {resistor_labels}")
print(f"Capacitor labels: {capacitor_labels}")

```

### **2.2. Node Voltage Maps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1780961535" width="250">



>* Match circuit node names to voltages
>* Use labels to reduce measurement ambiguity

>* Organize voltages from tests and diagnostics
>* Keep readings tied to circuit labels

>* Compare labeled voltages across operating conditions
>* Preserve circuit context for clearer summaries



```
#@title Python Code - Node Voltage Maps

# Dictionaries label circuit node voltages.
# Keys name nodes, values store readings.
# Engineers can summarize maps clearly.

# Create a node voltage dictionary.
node_voltages = {
    "input": 12.08,
    "regulator_output": 5.01,

    "feedback": 1.24,
    "sensor_line": 2.72,
    "ground": 0.00,
}

# Access important labeled voltages.
output_voltage = node_voltages["regulator_output"]
ground_voltage = node_voltages["ground"]
input_voltage = node_voltages["input"]

# Compute voltage differences safely.
load_drop = input_voltage - output_voltage
sensor_above_ground = node_voltages["sensor_line"] - ground_voltage
print("Node voltage map:", node_voltages)

# Print selected engineering interpretations.
print("Output voltage:", output_voltage, "V")
print("Input-to-output drop:", round(load_drop, 2), "V")
print("Sensor above ground:", round(sensor_above_ground, 2), "V")

# Define acceptable supply range.
minimum_output = 4.75
maximum_output = 5.25
output_ok = minimum_output <= output_voltage <= maximum_output

# Summarize pass or fail result.
print("Regulator output within tolerance:", output_ok)
print("Measured node count:", len(node_voltages))

# Aggregate values for a quick overview.
average_voltage = sum(node_voltages.values()) / len(node_voltages)
print("Average listed voltage:", round(average_voltage, 2), "V")

```

### **2.3. Measured Test Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780961536" width="250">



>* Label measurements by engineering meaning
>* Store mixed results as compact test records

>* Label test results from many engineering systems
>* Keep data readable, traceable, and unambiguous

>* Use keys to compare labeled test conditions
>* Link results to labels for clearer summaries



```
#@title Python Code - Measured Test Results

# Dictionaries label measured electrical test results.
# Keys describe measurements, not storage positions.
# Values can mix engineering data types.

# Store one compact bench test record.
test_result = {
    "serial_number": "PSU-042",
    "output_voltage_V": 5.08,
    "input_current_mA": 312.0,

    "ripple_mVpp": 18.5,
    "case_temperature_C": 41.2,
    "switching_frequency_kHz": 501.0,
    "passed_visual_check": True,
}

# Store specification limits using matching labels.
limits = {
    "output_voltage_V": (4.95, 5.05),
    "ripple_mVpp": (0.0, 25.0),
    "case_temperature_C": (0.0, 70.0),
}

# Validate that every required measurement exists.
required_keys = set(limits.keys())
measured_keys = set(test_result.keys())
missing_keys = required_keys - measured_keys
assert not missing_keys, "A required measured value is missing."

# Evaluate numeric limits and store pass results.
checks = {}
for name, bounds in limits.items():
    low_limit, high_limit = bounds
    measured_value = test_result[name]
    checks[name] = low_limit <= measured_value <= high_limit

# Add the final electrical decision to dictionary.
test_result["passed_electrical_limits"] = all(checks.values())
test_result["overall_pass"] = (
    test_result["passed_visual_check"]
    and test_result["passed_electrical_limits"]
)

# Print a short readable engineering summary.
print("Measured Test Results Dictionary")
print("Unit:", test_result["serial_number"])
print("Output voltage:", test_result["output_voltage_V"], "V")
print("Ripple:", test_result["ripple_mVpp"], "mV peak-to-peak")

# Iterate through labeled limit checks.
for name, did_pass in checks.items():
    status_text = "PASS" if did_pass else "FAIL"
    print(name, "limit check:", status_text)

# Show the final dictionary-based decision.
print("Overall pass:", test_result["overall_pass"])

```

## **3. Processing Signal Collections**

### **3.1. Signal Slicing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780961538" width="250">



>* Select important portions of signal samples
>* Focus analysis on the relevant time window

>* Sample positions map to signal time windows
>* Slice events for summaries and cleanup

>* Compare matching signal regions across tests
>* Choose slices carefully before summarizing data



```
#@title Python Code - Signal Slicing

# Signal slices isolate meaningful waveform regions.
# Engineers summarize selected samples, not everything.
# This example uses simple voltage samples.

# Store ordered oscilloscope voltage samples.
voltage_samples = [
    0.1, 0.2, 0.8, 2.4,

    5.6, 5.2, 5.0, 4.9,
    5.0, 5.1, 5.0, 5.0,
]

# Confirm enough samples exist before slicing.
expected_sample_count = 12
actual_sample_count = len(voltage_samples)
assert actual_sample_count == expected_sample_count

# Slice transient samples after switching begins.
transient_slice = voltage_samples[2:6]
transient_peak = max(transient_slice)
transient_average = sum(transient_slice) / len(transient_slice)

# Slice steady-state samples after settling.
steady_slice = voltage_samples[6:]
steady_average = sum(steady_slice) / len(steady_slice)
steady_minimum = min(steady_slice)

# Print compact engineering summaries only.
print("Total stored samples:", actual_sample_count)
print("Transient sample window:", transient_slice)
print("Transient peak voltage:", transient_peak, "V")
print("Transient average voltage:", round(transient_average, 2), "V")
print("Steady-state sample window:", steady_slice)
print("Steady-state average voltage:", round(steady_average, 2), "V")
print("Steady-state minimum voltage:", steady_minimum, "V")

```

### **3.2. Looping Signal Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780961541" width="250">



>* Loop samples to inspect signal behavior
>* Find spikes, noise, and useful patterns

>* Use sample values with their positions
>* Loop to check limits and patterns

>* Loop samples to build useful summaries
>* Turn raw signals into engineering decisions



```
#@title Python Code - Looping Signal Samples

# Loop through signal samples safely.
# Track positions, values, and summaries.
# Keep output short for learners.

# Store ordered voltage samples.
voltage_samples = [0.10, 1.20, 2.90, 5.40, 3.10, 1.00]

# Define an engineering safety threshold.
voltage_limit = 5.00

# Prepare counters and summary values.
violation_count = 0
total_voltage = 0.0

# Loop with index and value together.
for sample_index, voltage in enumerate(voltage_samples):
    total_voltage = total_voltage + voltage

    # Check each sample against the limit.
    if voltage > voltage_limit:
        violation_count = violation_count + 1

        # Report the flagged sample position.
        print(f"Sample {sample_index}: {voltage:.2f} V exceeds limit")

# Compute simple aggregate engineering summaries.
average_voltage = total_voltage / len(voltage_samples)
maximum_voltage = max(voltage_samples)

# Print concise results from looping.
print(f"Samples checked: {len(voltage_samples)}")
print(f"Average voltage: {average_voltage:.2f} V")
print(f"Maximum voltage: {maximum_voltage:.2f} V")
print(f"Limit violations: {violation_count}")

```

### **3.3. Batch Signal Summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780961540" width="250">



>* Summarize many samples into key signal values
>* Use summaries to make engineering decisions

>* Start with an engineering question.
>* Use iteration and aggregation for summaries.

>* Compare batches using consistent summary measures
>* Use summaries as guides, not replacements



```
#@title Python Code - Batch Signal Summaries

# Summarize batches of signal samples.
# Lists hold ordered voltage measurements.
# Aggregations support quick engineering decisions.

# Define two short voltage sample batches.
rail_a_volts = [4.98, 5.01, 5.07, 4.95, 5.10, 5.03]
rail_b_volts = [3.29, 3.31, 3.36, 3.28, 3.34, 3.30]

# Store labeled batches inside a dictionary.
signal_batches = {"5V rail": rail_a_volts, "3V3 rail": rail_b_volts}
safe_limits = {"5V rail": 5.05, "3V3 rail": 3.33}
summary_rows = []

# Process each labeled signal consistently.
for signal_name, samples in signal_batches.items():
    sample_count = len(samples)
    assert sample_count > 0, "Each batch needs samples"

    # Aggregate typical, extreme, and threshold values.
    average_value = sum(samples) / sample_count
    highest_value = max(samples)
    lowest_value = min(samples)

    # Count samples above the chosen safe limit.
    limit_value = safe_limits[signal_name]
    above_limit = sum(value > limit_value for value in samples)
    summary_rows.append((signal_name, average_value, highest_value, lowest_value, above_limit))

# Print compact engineering summaries.
print("Batch signal summaries:")
for row in summary_rows:
    name, avg, high, low, count = row
    print(f"{name}: avg={avg:.2f} V, max={high:.2f} V, min={low:.2f} V, above limit={count}")

# Use slicing to inspect a short time window.
window_samples = rail_a_volts[1:4]
window_average = sum(window_samples) / len(window_samples)
print(f"5V rail samples 1:4 average = {window_average:.2f} V")

```
