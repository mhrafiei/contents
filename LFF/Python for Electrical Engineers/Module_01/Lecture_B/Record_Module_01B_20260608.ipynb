### **1.3. Electrical Quantity Variables**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1780957724" width="250">



>* Store electrical quantities with clear variable names
>* Good names prevent confusion in circuit calculations

>* Choose numbers that match measurement precision
>* Include units in names to avoid ambiguity

>* Clear names connect code to measurements
>* Precise quantities prevent costly engineering misunderstandings



```
#@title Python Code - Electrical Quantity Variables

# Electrical quantities become clear named variables.
# Unit suffixes help prevent engineering mistakes.
# Simple arithmetic checks variable meaning.

# Store source voltage in volts.
source_voltage_volts = 12.0

# Store load resistance in ohms.
load_resistance_ohms = 240.0

# Calculate current using Ohm's law.
load_current_amperes = source_voltage_volts / load_resistance_ohms

# Store charge duration in seconds.
measurement_time_seconds = 60.0

# Calculate transferred charge in coulombs.
charge_coulombs = load_current_amperes * measurement_time_seconds

# Calculate load power in watts.
load_power_watts = source_voltage_volts * load_current_amperes

# Calculate energy used in joules.
energy_joules = load_power_watts * measurement_time_seconds

# Group results for compact display.
results = [
    ("Voltage", source_voltage_volts, "V"),
    ("Resistance", load_resistance_ohms, "ohm"),
    ("Current", load_current_amperes, "A"),
]

# Add charge, power, and energy.
results += [
    ("Charge", charge_coulombs, "C"),
    ("Power", load_power_watts, "W"),
    ("Energy", energy_joules, "J"),
]

# Print a readable engineering summary.
print("Electrical quantity variables:")

# Print each labeled value neatly.
for name, value, unit in results:
    print(f"{name}: {value:.3f} {unit}")

```

## **2. Engineering Expressions**

### **2.1. Ohms Law**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780957729" width="250">



>* Relates voltage, current, and resistance
>* Calculates unknown values for circuit checks

>* Choose the correct Ohm’s law expression
>* Check circuit values against real behavior

>* Use consistent units before calculating
>* Apply Ohm’s law to suitable components



```
#@title Python Code - Ohms Law

# Practice Ohms law with clear variables.
# Use arithmetic expressions for circuit quantities.
# Print compact results for quick checking.

# Define known voltage and resistance values.
voltage_volts = 12.0
resistance_ohms = 470.0
current_amps = voltage_volts / resistance_ohms

# Calculate voltage from current and resistance.
check_voltage_volts = current_amps * resistance_ohms
charge_coulombs = current_amps * 30.0
energy_joules = voltage_volts * charge_coulombs

# Calculate electrical power using Ohms law.
power_watts = voltage_volts * current_amps
current_milliamps = current_amps * 1000.0
resistance_kilohms = resistance_ohms / 1000.0

# Print values with engineering-style formatting.
print(f"Voltage: {voltage_volts:.2f} V")
print(f"Resistance: {resistance_kilohms:.3f} kOhm")
print(f"Current: {current_milliamps:.2f} mA")

# Print related charge, energy, and power.
print(f"Charge in 30 s: {charge_coulombs:.3f} C")
print(f"Energy used: {energy_joules:.3f} J")
print(f"Power dissipated: {power_watts:.3f} W")

```

### **2.2. Resistance Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780957726" width="250">



>* Resistance controls current flow in circuits
>* Expressions should match real circuit behavior

>* Series resistances add, parallel paths reduce resistance
>* Match expressions to actual circuit layout

>* Resistance guides design and troubleshooting decisions
>* Clear calculations support safe engineering judgment



```
#@title Python Code - Resistance Calculations

# Engineering expressions connect circuits and calculations.
# Resistance values reveal circuit current behavior.
# Clear variables make results easier checking.

# Store applied voltage and measured current.
voltage_volts = 12.0
current_amps = 0.025

# Calculate resistance using Ohm's law.
resistance_ohms = voltage_volts / current_amps
print(f"Single resistor: {resistance_ohms:.1f} ohms")

# Store three resistors connected in series.
r1_ohms = 220.0
r2_ohms = 330.0
r3_ohms = 470.0

# Add series resistances directly together.
series_total_ohms = r1_ohms + r2_ohms + r3_ohms
print(f"Series total: {series_total_ohms:.1f} ohms")

# Store two resistors connected in parallel.
parallel_a_ohms = 1000.0
parallel_b_ohms = 2000.0

# Calculate parallel resistance from reciprocal paths.
parallel_total_ohms = 1 / (1 / parallel_a_ohms + 1 / parallel_b_ohms)
print(f"Parallel total: {parallel_total_ohms:.1f} ohms")

# Estimate current through the series network.
supply_voltage_volts = 9.0
series_current_amps = supply_voltage_volts / series_total_ohms

# Print engineering current in milliamps.
series_current_mA = series_current_amps * 1000
print(f"Series current: {series_current_mA:.2f} mA")

# Compare resistance magnitudes using simple reasoning.
higher_than_parallel = series_total_ohms > parallel_total_ohms
print(f"Series exceeds parallel: {higher_than_parallel}")

```

### **2.3. Power Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780957730" width="250">



>* Power links circuit behavior to practical performance
>* Voltage and current calculate device power

>* Choose power formulas from known quantities
>* Apply expressions while preserving circuit meaning

>* Track units, order, and meaning carefully
>* Clear power expressions prevent engineering mistakes



```
#@title Python Code - Power Calculations

# Power calculations connect voltage and current.
# Clear variables make formulas easier to verify.
# Watts describe energy conversion per second.

# Store known circuit quantities using descriptive names.
voltage_volts = 12.0
current_amps = 0.75
resistance_ohms = 16.0

# Calculate power from voltage and current.
power_from_vi_watts = voltage_volts * current_amps

# Calculate power from current and resistance.
power_from_i2r_watts = current_amps ** 2 * resistance_ohms

# Calculate power from voltage and resistance.
power_from_v2r_watts = voltage_volts ** 2 / resistance_ohms

# Store operating time for energy estimation.
time_hours = 3.0
energy_watt_hours = power_from_vi_watts * time_hours

# Print compact engineering results.
print("Power calculation examples")
print("P = V * I =", power_from_vi_watts, "W")
print("P = I^2 * R =", power_from_i2r_watts, "W")
print("P = V^2 / R =", power_from_v2r_watts, "W")

# Print the related energy use.
print("Energy for", time_hours, "hours =", energy_watt_hours, "Wh")

# Check agreement between equivalent power expressions.
expressions_match = power_from_vi_watts == power_from_i2r_watts == power_from_v2r_watts
print("Equivalent expressions match:", expressions_match)

```

## **3. Readable Engineering Code**

### **3.1. Meaningful Comments**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780957732" width="250">



>* Explain why each engineering calculation exists
>* Help others review and improve code

>* Comments should explain assumptions, units, and intent
>* Clear comments help verify and catch errors

>* Keep comments clear, brief, and specific
>* Update comments to show engineering reasoning



```
#@title Python Code - Meaningful Comments

# Meaningful comments explain engineering intent.
# Values include units and assumptions.
# Readable code supports easier verification.

# Define nominal circuit values from design notes.
supply_voltage_volts = 12.0
load_resistance_ohms = 24.0
sensor_current_amps = 0.08

# Estimate load current using Ohm's law.
load_current_amps = supply_voltage_volts / load_resistance_ohms

# Check total current against regulator rating.
total_current_amps = load_current_amps + sensor_current_amps
regulator_limit_amps = 0.75

# Calculate load power for thermal review.
load_power_watts = supply_voltage_volts * load_current_amps
safe_power_watts = 8.0

# Store verification results using clear labels.
current_margin_amps = regulator_limit_amps - total_current_amps
power_margin_watts = safe_power_watts - load_power_watts

# Print concise results for design inspection.
print("Readable engineering code example")
print(f"Load current: {load_current_amps:.3f} A")
print(f"Total current: {total_current_amps:.3f} A")
print(f"Current margin: {current_margin_amps:.3f} A")

# Print thermal check with meaningful units.
print(f"Load power: {load_power_watts:.2f} W")
print(f"Power margin: {power_margin_watts:.2f} W")
print("Comments explain intent, assumptions, and limits.")

```

### **3.2. Inline Comments**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780957734" width="250">



>* Explain why each calculation matters
>* Link code values to real systems

>* Add context beyond clear variable names
>* Show units, sources, assumptions, and limits

>* Use inline comments sparingly for key reasoning
>* Show assumptions that support verification



```
#@title Python Code - Inline Comments

# Inline comments explain engineering intent.
# Clear units help verify calculations.
# Concise notes prevent crowded code.

# Store measured and design quantities.
supply_voltage_v = 12.0  # Measured bench supply voltage, in volts.
resistor_ohms = 220.0  # Standard resistor value, in ohms.
operating_time_s = 30.0  # Test duration used for energy estimate.

# Calculate current using Ohm's law.
current_a = supply_voltage_v / resistor_ohms  # Expected ideal current, amperes.
current_ma = current_a * 1000.0  # Convert amperes to milliamperes.

# Calculate power and energy values.
power_w = supply_voltage_v * current_a  # Resistor power dissipation, watts.
energy_j = power_w * operating_time_s  # Energy during the timed test, joules.

# Check against a conservative component limit.
power_limit_w = 0.25  # Quarter-watt resistor rating from datasheet.
is_safe = power_w <= power_limit_w  # Design passes if dissipation stays lower.

# Print a short verification summary.
print("Readable engineering code example")
print(f"Current: {current_ma:.2f} mA")
print(f"Power: {power_w:.3f} W")
print(f"Energy: {energy_j:.2f} J")
print(f"Within resistor rating: {is_safe}")

```

### **3.3. Code Comments**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780957735" width="250">



>* Comments explain engineering purpose and context
>* They help readers verify calculations later

>* Explain intent, not obvious code actions
>* Document units, sources, assumptions, and conditions

>* Comments make program logic easier to verify
>* Keep comments accurate as code changes



```
#@title Python Code - Code Comments

# Comments explain engineering intent and context.
# Clear notes support verification later.
# Units and assumptions prevent confusion.

# Store measured inputs with explicit engineering units.
voltage_v = 12.0
current_a = 0.50
resistance_ohm = 24.0

# Record the source and assumption for charge.
time_s = 10.0
charge_c = current_a * time_s
energy_j = voltage_v * charge_c

# Calculate power using Ohm law consistency.
power_w = voltage_v * current_a
expected_power_w = voltage_v**2 / resistance_ohm
power_difference_w = abs(power_w - expected_power_w)

# Print compact results for quick verification.
print("Readable Engineering Code: Code Comments")
print(f"Measured voltage: {voltage_v:.1f} V")
print(f"Measured current: {current_a:.2f} A")
print(f"Assumed resistance: {resistance_ohm:.1f} ohm")

# Show derived values with their units.
print(f"Charge over {time_s:.1f} s: {charge_c:.2f} C")
print(f"Energy delivered: {energy_j:.2f} J")
print(f"Power from V*I: {power_w:.2f} W")

# Compare two power formulas to check consistency.
print(f"Power check difference: {power_difference_w:.4f} W")

```
