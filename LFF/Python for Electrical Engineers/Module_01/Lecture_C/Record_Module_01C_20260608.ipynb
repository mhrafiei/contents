### **1.1. Rounding Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780958775" width="250">



>* Round engineering values for clearer communication
>* Keep rounded values usable in calculations

>* Match rounding precision to the context
>* Keep useful meaning without misleading detail

>* Round only when displaying final results
>* Use shared precision for clearer comparisons



```
#@title Python Code - Rounding Values

# This example teaches practical rounding choices.
# Engineering values often contain excessive decimals.
# Rounded numbers remain usable in calculations.

# Store several calculated electrical engineering values.
voltage_drop_volts = 1.234567
resistance_ohms = 47.89123
power_watts = voltage_drop_volts ** 2 / resistance_ohms

# Round values to context appropriate precision.
rounded_voltage = round(voltage_drop_volts, 2)
rounded_resistance = round(resistance_ohms, 1)
rounded_power = round(power_watts, 4)

# Show original and rounded numeric values.
print("Original voltage:", voltage_drop_volts)
print("Rounded voltage:", rounded_voltage)
print("Rounded resistance:", rounded_resistance)
print("Rounded power:", rounded_power)

# Round an energy estimate to a whole number.
energy_kwh = 1284.763
rounded_energy = round(energy_kwh)
print("Rounded site energy kWh:", rounded_energy)

# Verify rounded values are still numbers.
scaled_power_mw = rounded_power * 1000
print("Rounded power in mW:", scaled_power_mw)

```

### **1.2. Absolute Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1780958773" width="250">



>* Separate measurement size from direction
>* Use absolute value for consistent tolerances

>* Compare measurements by error size
>* Use magnitudes for clearer tolerance checks

>* Use magnitudes for reporting deviations clearly
>* Make threshold checks and decisions reliable



```
#@title Python Code - Absolute Values

# Absolute value reports distance from zero.
# Engineering signs can show measurement direction.
# Magnitudes help compare errors and tolerances.

# Store signed voltage readings in volts.
readings_volts = [-12.4, 3.2, -0.8, 9.6]

# Convert each signed value into magnitude.
magnitudes_volts = [abs(value) for value in readings_volts]

# Define expected voltage and measured value.
target_volts, measured_volts = 5.0, 4.73

# Compute signed error and absolute error.
signed_error = measured_volts - target_volts
absolute_error = abs(signed_error)

# Define allowable engineering tolerance.
tolerance_volts = 0.30
within_tolerance = absolute_error <= tolerance_volts

# Print compact inspection results.
print("Signed readings:", readings_volts)
print("Voltage magnitudes:", magnitudes_volts)
print("Signed error:", round(signed_error, 3), "V")
print("Absolute error:", round(absolute_error, 3), "V")
print("Within tolerance:", within_tolerance)

```

### **1.3. Measure Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1780958777" width="250">



>* Use length to count collection items
>* Check engineering data completeness before calculations

>* Check text lengths for valid entries
>* Catch missing or incorrectly sized data

>* Length counts items, not engineering magnitude
>* Convert text before numeric calculations



```
#@title Python Code - Measure Values

# Measure collections before trusting engineering data.
# Length checks reveal missing readings quickly.
# Text lengths validate codes and labels.

# Store a few voltage samples.
voltage_samples = [3.29, 3.31, 3.28, 3.30]

# Count readings using the len function.
sample_count = len(voltage_samples)

# Check an expected logger record count.
expected_count = 5

# Store engineering text labels.
part_code = "R100-EE7"
unit_label = "V rms"

# Measure text values using len.
code_length = len(part_code)
label_length = len(unit_label)

# Build a clear status message.
count_message = f"Samples stored: {sample_count} of {expected_count} expected"

# Build validation messages from lengths.
code_message = f"Part code characters: {code_length}"
label_message = f"Unit label characters: {label_length}"

# Compare actual and expected sizes.
missing_count = expected_count - sample_count

# Print compact measurement results.
print(count_message)
print(f"Missing samples: {missing_count}")
print(code_message)
print(label_message)
print(f"First reading type: {type(voltage_samples[0]).__name__}")

```

## **2. Math Library**

### **2.1. Trigonometric Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780958772" width="250">



>* Trigonometry models angles, rotations, and waveforms
>* Python supports practical electrical engineering calculations

>* Math trig functions expect radians
>* Convert degrees to avoid calculation errors

>* Resolve AC signals and phase relationships
>* Interpret trig results in engineering context



```
#@title Python Code - Trigonometric Calculations

# Demonstrate trigonometry for electrical phase calculations.
# Convert engineering angles before using functions.
# Print compact results for quick interpretation.

# Import the standard math library.
import math

# Store a measured phase angle in degrees.
phase_degrees = 30.0

# Convert degrees into radians for math functions.
phase_radians = math.radians(phase_degrees)

# Compute common trigonometric phase quantities.
sine_value = math.sin(phase_radians)
cosine_value = math.cos(phase_radians)
tangent_value = math.tan(phase_radians)

# Estimate power factor from the phase angle.
power_factor = cosine_value

# Use a small peak voltage example.
peak_voltage = 170.0
instant_voltage = peak_voltage * sine_value

# Display rounded engineering results clearly.
print(f"Phase angle: {phase_degrees:.1f} degrees")
print(f"Phase angle: {phase_radians:.4f} radians")
print(f"Sine component: {sine_value:.3f}")
print(f"Cosine component: {cosine_value:.3f}")
print(f"Tangent ratio: {tangent_value:.3f}")
print(f"Power factor estimate: {power_factor:.3f}")
print(f"Instantaneous voltage: {instant_voltage:.1f} V")

```

### **2.2. Trigonometric Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780958768" width="250">



>* Trig functions connect angles to measurements
>* Standard math tools ensure reusable calculations

>* Trig functions usually require radians
>* Convert degrees to avoid wrong results

>* Analyze phase, impedance, and signal timing
>* Convert angled measurements into reliable calculations



```
#@title Python Code - Trigonometric Functions

# Demonstrate trigonometric functions for electrical calculations.
# Convert familiar degree angles before using trig.
# Keep outputs short and engineering focused.

# Import the standard math library.
import math

# Store a phase angle in degrees.
phase_degrees = 30.0

# Convert degrees into required radians.
phase_radians = math.radians(phase_degrees)

# Compute sine and cosine components.
sine_value = math.sin(phase_radians)
cosine_value = math.cos(phase_radians)

# Compute tangent for impedance ratio.
tangent_value = math.tan(phase_radians)

# Use source voltage peak value.
voltage_peak = 170.0

# Estimate instantaneous voltage at phase.
instant_voltage = voltage_peak * sine_value

# Estimate resistive and reactive components.
resistance_ohms = 100.0
reactance_ohms = resistance_ohms * tangent_value

# Print compact engineering results.
print("Angle degrees:", round(phase_degrees, 2))
print("Angle radians:", round(phase_radians, 4))
print("sin(angle):", round(sine_value, 4))

# Print remaining compact results.
print("cos(angle):", round(cosine_value, 4))
print("tan(angle):", round(tangent_value, 4))
print("Instant voltage V:", round(instant_voltage, 2))

# Print impedance interpretation.
print("Reactive ohms for 100 ohm resistance:", round(reactance_ohms, 2))

```

### **2.3. Trigonometric Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780958770" width="250">



>* Model cycles using angles and phase
>* Convert degrees to radians for accuracy

>* Model waveform amplitude, timing, and phase
>* Relate phase angles to circuit behavior

>* Manage precision with rounding and tolerances
>* Interpret sign, scale, and engineering meaning



```
#@title Python Code - Trigonometric Calculations

# Demonstrate trigonometric waveform calculations.
# Convert degrees before using sine.
# Interpret phase shift engineering meaning.

# Import the standard math library.
import math

# Define waveform and angle values.
peak_voltage = 170.0
angle_degrees = 30.0
phase_degrees = -15.0

# Convert degree measures into radians.
angle_radians = math.radians(angle_degrees)
phase_radians = math.radians(phase_degrees)

# Calculate instantaneous sine wave voltage.
instant_voltage = peak_voltage * math.sin(angle_radians)
shifted_voltage = peak_voltage * math.sin(angle_radians + phase_radians)

# Calculate power factor from phase angle.
power_factor = math.cos(phase_radians)
phase_label = "lagging" if phase_degrees < 0 else "leading"

# Print compact engineering results.
print(f"Angle: {angle_degrees:.1f} degrees = {angle_radians:.3f} radians")
print(f"Unshifted voltage: {instant_voltage:.2f} V")
print(f"Shifted voltage: {shifted_voltage:.2f} V")
print(f"Phase: {abs(phase_degrees):.1f} degrees {phase_label}")
print(f"Power factor magnitude: {power_factor:.3f}")

```

## **3. Console Input Output**

### **3.1. Reading User Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780958779" width="250">



>* Input makes console calculations interactive
>* Supports repeated engineering checks with changing values

>* Convert typed text before doing calculations
>* Use clear prompts with expected units

>* Expect messy or unreasonable user entries
>* Use clear prompts and careful conversions



```
#@title Python Code - Reading User Input

# This example simulates console input safely.
# It shows text conversion before calculation.
# Prompts make engineering units more obvious.

# Store sample keyboard responses as text.
voltage_text = " 120.0 "
current_text = " 2.50 "
length_text = " 15 "

# Clean spaces before numeric conversion.
voltage_volts = float(voltage_text.strip())
current_amps = float(current_text.strip())
length_meters = float(length_text.strip())

# Calculate resistance and cable voltage drop.
load_resistance = voltage_volts / current_amps
cable_drop = 0.02 * length_meters * current_amps

# Display the simulated console conversation.
print("Enter voltage in volts:", voltage_text)
print("Enter current in amps:", current_text)
print("Enter cable length in meters:", length_text)

# Display formatted engineering results.
print(f"Load resistance: {load_resistance:.2f} ohms")
print(f"Cable voltage drop: {cable_drop:.2f} volts")
print("All numeric work used converted float values.")

```

### **3.2. Formatted Console Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780958780" width="250">



>* Show results clearly with units and precision
>* Prevent ambiguity in engineering calculator output

>* Control numeric precision for readable results
>* Add labels and units for clear interpretation

>* Clear output explains results and assumptions
>* Consistent formatting supports professional calculator use



```
#@title Python Code - Formatted Console Output

# Format results for engineering console output.
# Labels and units reduce ambiguity.
# Decimal places communicate practical precision.

# Store small example engineering values.
voltage_v = 12.0
resistance_ohm = 47.0
current_a = voltage_v / resistance_ohm

# Convert amperes into milliamperes.
current_ma = current_a * 1000.0
power_w = voltage_v * current_a

# Print a compact labeled report.
print("DC circuit calculation")
print(f"Voltage:    {voltage_v:8.2f} V")
print(f"Resistance: {resistance_ohm:8.2f} ohm")

# Align related values for quick comparison.
print(f"Current:    {current_ma:8.2f} mA")
print(f"Power:      {power_w:8.3f} W")

# Show scientific notation for tiny values.
capacitance_f = 0.00000047
print(f"Capacitance: {capacitance_f:.2e} F")

```

### **3.3. Formatted Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Electrical Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780958782" width="250">



>* Label results with units and precision
>* Clear formatting reduces engineering calculation ambiguity

>* Round values to meaningful precision
>* Show clear labels, units, and conventions

>* Clear output makes calculators easier to follow
>* Good formatting improves checking and reliability



```
#@title Python Code - Formatted Output

# This script demonstrates formatted engineering output.
# Values are labeled with units and precision.
# F strings make console results readable.

# Store sample calculator input values.
voltage_volts = 12.0
resistance_ohms = 47.0
frequency_hz = 1000.0

# Calculate common electrical quantities.
current_amps = voltage_volts / resistance_ohms
power_watts = voltage_volts * current_amps
period_seconds = 1.0 / frequency_hz

# Print a clear calculator-style heading.
print("Formatted Output: DC Circuit Calculator")

# Show input values with units.
print(f"Voltage: {voltage_volts:.2f} V")
print(f"Resistance: {resistance_ohms:.1f} ohms")

# Show calculated results with controlled precision.
print(f"Current: {current_amps:.3f} A")
print(f"Power: {power_watts:.2f} W")

# Show scientific notation for small timing values.
print(f"Period: {period_seconds:.3e} s")

# Build one formatted summary line.
summary_text = f"At {frequency_hz:.0f} Hz, period is {period_seconds*1000:.3f} ms."
print(summary_text)

```
