## **1. Ordered Medical Readings**

### **1.1. Patient Data Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_01_01.jpg?v=1781141711" width="250">



>* Lists store ordered patient readings
>* Sequences reveal trends and support care

>* Lists organize ordered nonnumeric patient information
>* Use lists when sequence has clinical meaning

>* Keep list items consistent and comparable
>* Handle patient information accurately and ethically



```
#@title Python Code - Patient Data Lists

# Lists store ordered patient measurements.
# Each value keeps chronological clinical meaning.
# Small examples protect sensitive patient details.

# Create ordered morning glucose readings.
glucose_readings = [118, 124, 121, 116]

# Add the newest reading last.
glucose_readings.append(113)

# Retrieve baseline and latest values.
baseline_reading = glucose_readings[0]
latest_reading = glucose_readings[-1]

# Replace a corrected charting value.
glucose_readings[1] = 122

# Count readings and compute simple change.
reading_count = len(glucose_readings)
latest_change = latest_reading - baseline_reading

# Show the ordered clinical sequence.
print("Ordered glucose readings:", glucose_readings)

# Show baseline and most recent values.
print("Baseline reading:", baseline_reading, "mg/dL")
print("Latest reading:", latest_reading, "mg/dL")

# Show whether values decreased overall.
print("Change from baseline:", latest_change, "mg/dL")
print("Number of readings:", reading_count)

# Summarize why a list fits.
print("A list fits because order matters.")

```

### **1.2. Tuple Vital Signs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_01_02.jpg?v=1781141709" width="250">



>* Tuples group vital signs in fixed order
>* Each tuple represents one complete observation

>* Tuples preserve linked vital signs in order
>* Grouped readings show one clinical moment

>* Use tuples for fixed vital-sign patterns
>* Choose lists for growing observation histories



```
#@title Python Code - Tuple Vital Signs

# Tuples can store fixed clinical snapshots.
# Vital signs often have ordered meaning.
# This example indexes tuple values safely.

# Create one ordered vital-sign tuple.
vital_signs = (37.8, 104, 22, 128, 76, 95)

# Store matching labels in another tuple.
vital_labels = ("temperature_C", "pulse_bpm", "resp_rate", "systolic", "diastolic", "oxygen_sat")

# Check both tuples share expected length.
expected_count = 6
actual_count = len(vital_signs)

# Print a compact summary for learners.
print("Vital sign tuple:", vital_signs)
print("Number of readings:", actual_count)

# Retrieve values by stable positions.
temperature_c = vital_signs[0]
pulse_bpm = vital_signs[1]
oxygen_sat = vital_signs[-1]

# Show how positions carry meaning.
print("Temperature C:", temperature_c)
print("Pulse bpm:", pulse_bpm)
print("Oxygen saturation %:", oxygen_sat)

# Pair labels and values for readable output.
labeled_readings = tuple(zip(vital_labels, vital_signs))
print("First labeled reading:", labeled_readings[0])

# Confirm the fixed structure before use.
structure_ok = actual_count == expected_count
print("Expected tuple structure:", structure_ok)

```

### **1.3. Patient Reading Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_01_03.jpg?v=1781141713" width="250">



>* Store one patient’s readings in order
>* Show trends across repeated medical measurements

>* Lists capture repeated readings across care settings
>* Order gives simple values clinical context

>* Keep readings consistent and comparable
>* Ordered lists preserve the clinical story



```
#@title Python Code - Patient Reading Lists

# Lists keep repeated patient readings ordered.
# Each value represents one timed observation.
# Consistent units make comparisons safer.

# Store morning glucose readings in mg/dL.
glucose_readings = [118, 124, 121, 130, 126]

# Store temperature readings in degrees Celsius.
temperature_readings = [37.1, 37.4, 37.2, 37.8]

# Retrieve readings by their ordered positions.
first_glucose = glucose_readings[0]
latest_glucose = glucose_readings[-1]

# Add a new reading at the end.
glucose_readings.append(122)
updated_count = len(glucose_readings)

# Calculate a simple average from the list.
average_glucose = sum(glucose_readings) / updated_count

# Print concise patient list summaries.
print("Glucose readings:", glucose_readings)
print("First glucose reading:", first_glucose)
print("Latest glucose reading:", latest_glucose)
print("Number of glucose readings:", updated_count)
print("Average glucose reading:", round(average_glucose, 1))
print("Temperature readings:", temperature_readings)

```

## **2. Patient Data Lookup**

### **2.1. Updating Patient Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_02_01.jpg?v=1781141705" width="250">



>* Patient details often need careful updates
>* Change only the correct record field

>* Use keys for labeled patient details
>* Use indexes carefully; keep tuples stable

>* Update records accurately and consistently
>* Use collections to preserve patient meaning



```
#@title Python Code - Updating Patient Records

# Patient dictionaries make updates readable.
# Lists preserve ordered patient observations.
# Sets prevent duplicated allergy labels.

# Create one compact patient record.
patient_record = {
    "name": "Jordan Lee",
    "phone": "555-0101",
    "allergies": {"penicillin"},
    "blood_pressure": ["128/82", "124/80"],
}

# Display selected values before updating.
print("Before:", patient_record["phone"])
print("Before allergies:", sorted(patient_record["allergies"]))

# Update one dictionary value by key.
patient_record["phone"] = "555-0199"

# Add a new ordered blood pressure reading.
patient_record["blood_pressure"].append("122/78")

# Add allergy labels without duplicates.
patient_record["allergies"].add("latex")
patient_record["allergies"].add("penicillin")

# Store a fixed visit date tuple.
visit_date = (2026, 6, 11)

# Display focused results after updating.
print("After:", patient_record["phone"])
print("Latest BP:", patient_record["blood_pressure"][-1])
print("Allergies:", sorted(patient_record["allergies"]))
print("Visit date tuple:", visit_date)

```

### **2.2. Patient Key Access**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_02_02.jpg?v=1781141704" width="250">



>* Use keys to find named patient values
>* Labels improve readability, safety, and maintenance

>* Update changing patient fields directly by key
>* Organize values with clinically meaningful labels

>* Use consistent keys and handle missing data
>* Clear labels support safer patient lookups



```
#@title Python Code - Patient Key Access

# Dictionaries support patient key access.
# Keys retrieve clinically meaningful values.
# Updates target one named field.

# Create a small patient record.
patient_record = {
    "patient_id": "M1042",
    "blood_type": "O+",
    "allergies": ["penicillin"],
    "room": "12A",
}

# Retrieve values using keys.
patient_id_value = patient_record["patient_id"]
allergy_list_value = patient_record["allergies"]
print("Patient ID:", patient_id_value)

# Check membership inside the allergy list.
has_penicillin_allergy = "penicillin" in allergy_list_value
print("Penicillin allergy:", has_penicillin_allergy)

# Update one named field safely.
patient_record["room"] = "14B"
print("Updated room:", patient_record["room"])

# Add a new key after results arrive.
patient_record["latest_glucose_mg_dL"] = 118
print("Glucose result:", patient_record["latest_glucose_mg_dL"])

# Use get for possibly missing keys.
follow_up_value = patient_record.get("follow_up", "not documented")
print("Follow-up status:", follow_up_value)

# Organize selected fields for a summary.
summary_keys = ["patient_id", "room", "blood_type"]
summary_values = {key: patient_record[key] for key in summary_keys}
print("Summary fields:", summary_values)

```

### **2.3. List Indexing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_02_03.jpg?v=1781141707" width="250">



>* Use positions to find ordered patient values
>* Check list order to avoid mistakes

>* Update ordered patient values by position
>* Use keys for mixed patient details

>* Use indexing to compare ordered patient trends
>* Confirm list order before trusting positions



```
#@title Python Code - List Indexing

# Practice list indexing with patient readings.
# Positions help retrieve ordered clinical values.
# Updates can correct documented patient data.

# Store ordered oxygen saturation readings.
oxygen_saturations = [94, 95, 93, 96]

# Confirm enough readings exist before indexing.
if len(oxygen_saturations) >= 4:
    first_reading = oxygen_saturations[0]
    latest_reading = oxygen_saturations[-1]
else:
    first_reading = None
    latest_reading = None

# Correct the third recorded reading.
oxygen_saturations[2] = 94

# Add a new reassessment reading.
oxygen_saturations.append(97)

# Retrieve specific values using positions.
third_reading = oxygen_saturations[2]
latest_reading = oxygen_saturations[-1]

# Print a concise patient-data summary.
print("Oxygen saturation readings:", oxygen_saturations)
print("Admission reading:", first_reading, "%")
print("Corrected third reading:", third_reading, "%")
print("Most recent reading:", latest_reading, "%")
print("Total recorded readings:", len(oxygen_saturations))

```

## **3. Collection Methods**

### **3.1. Append Patient Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_03_01.jpg?v=1781141714" width="250">



>* Lists grow as new clinical data arrives
>* Order helps reveal patient changes over time

>* Append when past values still matter
>* Use keyed structures for replacements

>* Match append use to clinical data purpose
>* Use ordered lists for reviewable patient events



```
#@title Python Code - Append Patient Values

# Lists can grow as observations arrive.
# Appending preserves earlier patient values.
# This example tracks heart rates.

# Start with admission heart rate.
heart_rates = [88]

# Append values from later checks.
heart_rates.append(92)
heart_rates.append(86)
heart_rates.append(84)

# Store richer events as dictionaries.
symptom_log = []
symptom_log.append({"time": "08:00", "symptom": "nausea"})

# Add another patient-reported symptom.
symptom_log.append({"time": "10:30", "symptom": "dizziness"})

# Print the updated ordered values.
print("Heart rate timeline:", heart_rates)
print("Most recent heart rate:", heart_rates[-1])
print("Symptom event count:", len(symptom_log))
print("Latest symptom:", symptom_log[-1])

```

### **3.2. Updating Collections**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_03_02.jpg?v=1781141716" width="250">



>* Patient data updates require suitable collections
>* Lists, dictionaries, sets, and tuples differ

>* Match collection type to update needs
>* Use labels to prevent clinical confusion

>* Update data with clinical meaning in mind
>* Choose collections based on needed change history



```
#@title Python Code - Updating Collections

# Collections support safe clinical updates.
# Choose structures based on task meaning.
# This example updates patient information.

# Store ordered temperature readings.
temperatures_c = [37.1, 37.4, 38.0]

# Add a new measurement.
temperatures_c.append(37.8)

# Correct an entered value.
temperatures_c[2] = 37.9

# Store labeled vital signs.
vitals = {"heart_rate": 88, "systolic_bp": 132}

# Update a value by key.
vitals["heart_rate"] = 84

# Add a newly measured value.
vitals["oxygen_percent"] = 97

# Store unique allergy labels.
allergies = {"penicillin", "latex"}

# Add and remove set members.
allergies.add("iodine")
allergies.discard("latex")

# Store stable birth date values.
birth_date = (1980, 5, 14)

# Print concise clinical summaries.
print("Updated temperatures:", temperatures_c)
print("Updated vitals:", vitals)
print("Current allergies:", sorted(allergies))
print("Stable birth date:", birth_date)
print("List keeps order; dictionary uses labels.")
print("Set prevents duplicates; tuple stays unchanged.")

```

### **3.3. Set Methods**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_A/image_03_03.jpg?v=1781141718" width="250">



>* Use sets for unique medical items
>* Check membership without order or duplicates

>* Add or remove unique patient risk labels
>* Use sets for checklist-like medical data

>* Compare medical groups for overlap or differences
>* Use sets for distinct membership relationships



```
#@title Python Code - Set Methods

# Sets store unique clinical labels.
# Methods update membership safely.
# Comparisons reveal group relationships.

# Create allergy entries with repeated labels.
allergy_entries = ["penicillin", "latex", "penicillin", "iodine"]

# Convert the list into unique allergies.
unique_allergies = set(allergy_entries)
print("Unique allergies:", sorted(unique_allergies))

# Add a newly confirmed allergy label.
unique_allergies.add("sulfa")
print("After add:", sorted(unique_allergies))

# Remove a ruled-out allergy safely.
unique_allergies.discard("iodine")
print("After discard:", sorted(unique_allergies))

# Compare medications before and after admission.
before_admission = {"metformin", "lisinopril", "atorvastatin"}
after_discharge = {"metformin", "apixaban", "atorvastatin"}

# Find shared, stopped, and started medicines.
continued = before_admission.intersection(after_discharge)
stopped = before_admission.difference(after_discharge)
started = after_discharge.difference(before_admission)

# Print comparison results concisely.
print("Continued medicines:", sorted(continued))
print("Stopped medicines:", sorted(stopped))
print("Started medicines:", sorted(started))

# Use union for overall medication review.
full_review = before_admission.union(after_discharge)
print("All medicines reviewed:", sorted(full_review))

```
