## **1. Clinical Text Cleaning**

### **1.1. Cleaning Clinical Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_01_01.jpg?v=1781142622" width="250">



>* Standardize clinical labels without changing meaning
>* Remove text variations for reliable interpretation

>* Standardize spacing, capitalization, and separators
>* Preserve clinically important punctuation and units

>* Match abbreviations to audience and context
>* Reduce noise without losing clinical meaning



```
#@title Python Code - Cleaning Clinical Labels

# Clinical labels often arrive messy.
# Simple string tools reduce variation.
# Cleaning must preserve clinical meaning.

import re

# Store several short clinical labels.
raw_labels = [
    "  heart   rate  ",
    "Blood_Pressure",
    "Vitals: BP",
    "PATIENT.ID",
]

# Define a cautious label cleaner.
def clean_label(label):
    text = label.strip()
    text = re.sub(r"[_\-.]+", " ", text)
    text = re.sub(r"\s+", " ", text)
    text = text.replace("Vitals: ", "")
    text = text.lower()
    return text.capitalize()

# Clean each label using comprehension.
cleaned_labels = [clean_label(label) for label in raw_labels]

# Pair original and cleaned labels.
paired_labels = list(zip(raw_labels, cleaned_labels))

# Display compact cleaning results.
print("Clinical label cleaning examples:")
for original, cleaned in paired_labels:
    print(f"{original!r} -> {cleaned!r}")

# Show why clinical meaning matters.
print("Keep units distinct: mg is not mcg.")

```

### **1.2. Cleaning Clinical Notes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_01_02.jpg?v=1781142623" width="250">



>* Clinical notes often contain inconsistent text
>* Clean formatting while preserving clinical meaning

>* Preserve meaning during note cleanup
>* Protect clinical details while improving readability

>* Match cleaning to note source and purpose
>* Keep originals and protect clinical meaning



```
#@title Python Code - Cleaning Clinical Notes

# Clean clinical notes carefully.
# Preserve meaning during text cleaning.
# Keep originals beside cleaned versions.

# Import regular expressions for text cleaning.
import re

# Store tiny example clinical notes.
notes = [
    "  Pt   denies chest pain.  SOB??  ",
    "BP 140/90 -- take 1 tablet twice daily.",
    "Patient walks 50 feet; wound is 2 cm."
]

# Define conservative cleaning behavior.
def clean_note(note):
    """Return a readable note without removing clinical meaning."""
    cleaned = note.replace("\n", " ")
    cleaned = re.sub(r"[-]{2,}", " - ", cleaned)
    cleaned = re.sub(r"\s+", " ", cleaned).strip()
    return cleaned

# Clean each note while preserving originals.
cleaned_notes = [clean_note(note) for note in notes]

# Show original and cleaned examples.
print("Clinical note cleaning examples:")
for index, pair in enumerate(zip(notes, cleaned_notes), start=1):
    original, cleaned = pair
    print(f"{index}. Original: {original!r}")
    print(f"   Cleaned:  {cleaned!r}")

# Check that negation words were preserved.
negation_preserved = "denies" in cleaned_notes[0].lower()

# Print a brief safety reminder.
print(f"Negation preserved: {negation_preserved}")

```

### **1.3. Cleaning Clinical Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_01_03.jpg?v=1781142625" width="250">



>* Clinical labels identify medical concepts from many sources
>* Clean labels while preserving clinical meaning

>* Standardize spaces, capitalization, units, and labels
>* Handle punctuation and abbreviations with context

>* Standardize labels for their intended purpose
>* Keep originals to preserve clinical meaning



```
#@title Python Code - Cleaning Clinical Labels

# Clinical labels often contain small inconsistent variations.
# Built-in strings help standardize simple medical labels.
# Keep original labels for traceable clinical cleaning.

# Import regular expressions for repeated spacing.
import re

# Store short labels from different clinical systems.
raw_labels = [
    "  blood pressure  ",
    "B.P.",

    "Diabetes,, Type 2",
    "paracetamol 500MG",
    "  icu ward  ",
]

# Map known synonyms using careful local rules.
synonym_map = {
    "b.p.": "Blood Pressure",
    "icu ward": "Intensive Care Unit Ward",
}

# Clean one label while preserving clinical meaning.
def clean_label(label):
    # Remove edge spaces and repeated internal spaces.
    text = re.sub(r"\s+", " ", label.strip())

    # Remove repeated commas without touching meaningful hyphens.
    text = re.sub(r",+", ",", text)

    # Standardize selected unit capitalization carefully.
    text = text.replace("500MG", "500 mg")

    # Apply trusted synonym mapping before title casing.
    mapped = synonym_map.get(text.lower(), text)

    # Use title case for simple readable labels.
    cleaned = mapped.title()
    return cleaned

# Build paired original and cleaned labels.
cleaned_pairs = [(label, clean_label(label)) for label in raw_labels]

# Print a compact traceable cleaning table.
print("Original label -> cleaned label")

# Show only a few short teaching examples.
for original, cleaned in cleaned_pairs:
    print(f"{original!r} -> {cleaned!r}")

# Summarize why traceability matters.
print(f"Kept {len(cleaned_pairs)} original labels for review.")

```

## **2. Basic Medical Files**

### **2.1. Reading Medical Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_02_01.jpg?v=1781142616" width="250">



>* Identify file type, location, and encoding
>* Read accurately to preserve clinical meaning

>* Choose whole-file or line-by-line reading
>* Validate each line before trusting it

>* Respect CSV structure, quotes, and missing values
>* Convert values carefully and protect confidentiality



```
#@title Python Code - Reading Medical Files

# This example reads small medical files.
# It demonstrates safe built-in file handling.
# The output stays short and readable.

from pathlib import Path
import csv
from datetime import datetime

# Create a safe temporary lesson folder.
folder = Path("medical_file_lesson")
folder.mkdir(exist_ok=True)

# Define a tiny clinical text file.
text_path = folder / "discharge_note.txt"
note_lines = ["Drink fluids after discharge.\n", "Call clinic if fever returns.\n"]

# Write the text file for reading practice.
with text_path.open("w", encoding="utf-8") as file:
    file.writelines(note_lines)

# Read the text file line by line.
clean_notes = []
with text_path.open("r", encoding="utf-8") as file:
    for line in file:
        clean_notes.append(line.strip())

# Define a tiny CSV-style observation file.
csv_path = folder / "vital_signs.csv"
rows = [
    ["patient_id", "visit_date", "systolic", "diastolic", "note"],
    ["P001", "2026-06-01", "128", "78", "Stable, review in clinic"],
    ["P002", "2026-06-02", "142", "88", "Repeat pressure check"],
]

# Write rows using csv to preserve commas.
with csv_path.open("w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Read structured rows safely with DictReader.
records = []
with csv_path.open("r", encoding="utf-8", newline="") as file:
    reader = csv.DictReader(file)
    for row in reader:
        row["visit_date"] = datetime.fromisoformat(row["visit_date"]).date()
        records.append(row)

# Print a compact teaching summary.
print("Text lines read:", len(clean_notes))
print("First instruction:", clean_notes[0])
print("CSV records read:", len(records))
print("First patient:", records[0]["patient_id"])
print("First note:", records[0]["note"])
print("First visit date:", records[0]["visit_date"])
print("Files stored in:", folder.resolve().name)

```

### **2.2. Writing Medical Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_02_02.jpg?v=1781142618" width="250">



>* Save program data for later use
>* Use clear, consistent medical file structure

>* Plain text suits narrative medical information
>* CSV files need consistent, careful formatting

>* Choose private, purposeful file names and locations
>* Prevent overwrites; verify readable, durable saved data



```
#@title Python Code - Writing Medical Files

# Practice writing simple medical files.
# Use safe paths and tiny records.
# Verify saved text after writing.

# Import standard file handling helpers.
from pathlib import Path
import csv
from datetime import date

# Create a safe teaching folder.
folder = Path("medical_file_examples")
folder.mkdir(exist_ok=True)
report_date = date.today().isoformat()

# Prepare a plain text instruction file.
text_path = folder / f"discharge_instructions_{report_date}.txt"
instructions = [
    "Drink water regularly after the visit.",
    "Call the clinic for fever above 100.4 F.",
    "Walk short distances as tolerated today.",
]

# Write the plain text instructions.
with text_path.open("w", encoding="utf-8") as file:
    file.write("Basic discharge instructions\n")
    file.write("\n".join(instructions))

# Prepare small CSV style records.
csv_path = folder / f"followup_calls_{report_date}.csv"
rows = [
    {"patient_id": "A001", "date": report_date, "status": "scheduled"},
    {"patient_id": "A002", "date": report_date, "status": "completed"},
]

# Write rows with safe CSV handling.
fieldnames = ["patient_id", "date", "status"]
with csv_path.open("w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

# Reopen files to confirm writing worked.
text_preview = text_path.read_text(encoding="utf-8").splitlines()[:2]
with csv_path.open("r", newline="", encoding="utf-8") as file:
    csv_preview = list(csv.DictReader(file))[:1]

# Print concise verification results.
print("Created files:", text_path.name, "and", csv_path.name)
print("Text preview:", text_preview)
print("CSV first row:", csv_preview[0])

```

### **2.3. Updating Medical Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_02_03.jpg?v=1781142620" width="250">



>* Preserve existing data during medical file updates
>* Make changes carefully and verify results

>* Append safely; modify records carefully
>* Use clear structure and reliable identifiers

>* Back up files before making updates
>* Review changes to confirm accuracy



```
#@title Python Code - Updating Medical Files

# Practice safe updates for medical files.
# Keep backups before changing records.
# Verify results after writing changes.

from pathlib import Path
import csv
from datetime import date

# Create a small teaching data folder.
data_folder = Path("medical_file_demo")
data_folder.mkdir(exist_ok=True)

# Define original and backup file paths.
csv_path = data_folder / "vaccination_schedule.csv"
backup_path = data_folder / "vaccination_schedule_backup.csv"

# Prepare tiny anonymized appointment rows.
rows = [
    {"study_id": "A102", "status": "arrived", "note": "ok"},
    {"study_id": "B205", "status": "pending", "note": "call"},
    {"study_id": "C309", "status": "pending", "note": "reschedule"},
]

# Write the starting CSV-style medical file.
with csv_path.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=rows[0].keys())
    writer.writeheader()
    writer.writerows(rows)

# Save a backup before making updates.
backup_path.write_text(csv_path.read_text())

# Read all rows before modifying one record.
with csv_path.open("r", newline="") as file:
    current_rows = list(csv.DictReader(file))

# Update one record using a clear identifier.
for row in current_rows:
    if row["study_id"] == "B205":
        row["status"] = "completed"
        row["note"] = f"updated {date.today().isoformat()}"

# Rewrite the file after controlled changes.
with csv_path.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=current_rows[0].keys())
    writer.writeheader()
    writer.writerows(current_rows)

# Append an audit entry to a plain text file.
audit_path = data_folder / "update_audit_log.txt"
audit_line = f"{date.today().isoformat()} updated B205 status\n"
audit_path.open("a").write(audit_line)

# Reopen the updated file for verification.
with csv_path.open("r", newline="") as file:
    verified_rows = list(csv.DictReader(file))

# Print a short verification summary.
changed = [row for row in verified_rows if row["study_id"] == "B205"]
print("Backup created:", backup_path.exists())
print("Updated record:", changed[0])
print("Audit entries:", len(audit_path.read_text().splitlines()))

```

## **3. Standard Modules**

### **3.1. Date Handling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_03_01.jpg?v=1781142608" width="250">



>* Medical dates carry important clinical meaning
>* Structured dates enable reliable comparisons and calculations

>* Convert varied date text into structured values
>* Format dates consistently to preserve meaning

>* Use date tools for clinical timeline calculations
>* Handle timestamps, leap years, and time zones



```
#@title Python Code - Date Handling

# Dates make clinical timelines safer.
# Structured dates prevent ambiguous comparisons.
# This example parses medical date text.

from datetime import datetime, date
from pathlib import Path
import csv

# Store a tiny CSV safely in Colab.
folder_path = Path("medical_dates_demo")
folder_path.mkdir(exist_ok=True)

# Keep source rows small and readable.
records = [
    {"patient_id": "P001", "dob": "1980-04-03", "discharge": "2026-03-01"},
    {"patient_id": "P002", "dob": "1975-11-20", "discharge": "2026-03-05"},
]

# Write the small file with headers.
file_path = folder_path / "patient_dates.csv"
with file_path.open("w", newline="") as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=records[0].keys())
    writer.writeheader()
    writer.writerows(records)

# Parse text dates into structured date objects.
today = date(2026, 3, 10)
parsed_rows = []

# Read rows back from the saved file.
with file_path.open("r", newline="") as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        birth_date = datetime.strptime(row["dob"], "%Y-%m-%d").date()
        discharge_date = datetime.strptime(row["discharge"], "%Y-%m-%d").date()
        days_since = (today - discharge_date).days
        age_years = today.year - birth_date.year
        age_years -= (today.month, today.day) < (birth_date.month, birth_date.day)
        parsed_rows.append((row["patient_id"], age_years, days_since))

# Display concise structured results.
print("Date handling summary:")
for patient_id, age_years, days_since in parsed_rows:
    print(f"{patient_id}: age {age_years}, discharged {days_since} days ago")

# Format one date consistently for reporting.
report_date = today.isoformat()
print(f"Report date in ISO format: {report_date}")

```

### **3.2. Safe Paths**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_03_02.jpg?v=1781142610" width="250">



>* Handle medical file locations safely
>* Build paths from clear structured parts

>* Protect sensitive data with deliberate file locations
>* Separate folders prevent overwrites and aid reproducibility

>* Check user file names for unsafe paths
>* Use path tools to protect medical files



```
#@title Python Code - Safe Paths

# Safe paths protect medical text files.
# Pathlib builds locations without fragile strings.
# This example writes tiny clinic exports.

# Import standard modules for safe paths.
from pathlib import Path
from datetime import date
import csv

# Create a controlled project workspace.
project_dir = Path.cwd() / "safe_path_demo"
input_dir = project_dir / "raw_exports"
output_dir = project_dir / "review_outputs"

# Create folders deliberately before writing.
input_dir.mkdir(parents=True, exist_ok=True)
output_dir.mkdir(parents=True, exist_ok=True)
today_text = date.today().isoformat()

# Build a safe input file path.
raw_file = input_dir / f"clinic_notes_{today_text}.csv"
rows = [["patient_id", "label"], ["A001", " BP Follow Up "]]

# Write a tiny CSV-style export.
with raw_file.open("w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Read back and clean labels.
with raw_file.open("r", newline="", encoding="utf-8") as file:
    reader = list(csv.reader(file))
    cleaned_label = reader[1][1].strip().lower().replace(" ", "_")

# Prevent escaping the approved folder.
requested_name = "cleaned_labels.txt"
safe_file = (output_dir / requested_name).resolve()
approved_dir = output_dir.resolve()

# Confirm output stays inside folder.
if approved_dir not in safe_file.parents:
    raise ValueError("Unsafe output location blocked")

# Write a small reviewed output file.
message = f"{reader[1][0]},{cleaned_label}\n"
safe_file.write_text(message, encoding="utf-8")
output_text = safe_file.read_text(encoding="utf-8").strip()

# Print concise teaching results.
print("Project folder:", project_dir.name)
print("Raw file exists:", raw_file.is_file())
print("Output file exists:", safe_file.is_file())
print("Cleaned record:", output_text)

```

### **3.3. Structured Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_02/Lecture_B/image_03_03.jpg?v=1781142612" width="250">



>* Organize medical information into clear fields
>* Consistent formats make later processing safer

>* Use CSV rows and fields consistently
>* Preserve messy medical data across systems

>* Reliable structure supports accountability and reuse
>* Built-in tools create portable summary records



```
#@title Python Code - Structured Output

# Structured output preserves medical record meaning.
# Standard modules safely write simple files.
# Small examples help beginners inspect results.

import csv
import json
from pathlib import Path
from datetime import date

# Create a safe Colab output folder.
output_folder = Path("medical_outputs")
output_folder.mkdir(exist_ok=True)

# Define two tiny structured reminder records.
reminders = [
    {
        "patient_id": "P001",
        "reminder_date": date.today().isoformat(),
        "instruction": "Take amoxicillin, 500 mg twice daily.",
        "follow_up": "Call clinic in 48 hours.",
    },

    {
        "patient_id": "P002",
        "reminder_date": date.today().isoformat(),
        "instruction": "Check blood pressure each morning.",
        "follow_up": "Bring log to next visit.",
    },
]

# Choose stable field order for CSV output.
fieldnames = [
    "patient_id",
    "reminder_date",
    "instruction",
    "follow_up",
]

# Write comma-safe structured rows using csv.
csv_path = output_folder / "appointment_reminders.csv"
with csv_path.open("w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(reminders)

# Write the same records as readable JSON.
json_path = output_folder / "appointment_reminders.json"
with json_path.open("w", encoding="utf-8") as file:
    json.dump(reminders, file, indent=2)

# Read back the CSV to verify structure.
with csv_path.open("r", newline="", encoding="utf-8") as file:
    rows = list(csv.DictReader(file))

# Print a short, readable summary.
print("Structured output example completed.")
print(f"CSV file: {csv_path}")
print(f"JSON file: {json_path}")
print(f"Records written and verified: {len(rows)}")
print("First instruction:", rows[0]["instruction"])

```
