## **1. Clinical Conditionals**

### **1.1. Threshold Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_01_01.jpg?v=1781140726" width="250">



>* Compare medical values with clinical thresholds
>* Use conditionals to label results consistently

>* Define range boundaries precisely
>* Order conditions for safe classification

>* Programs classify; clinicians interpret context
>* Practice builds dependable clinical decision logic



```
#@title Python Code - Threshold Classification

# Clinical thresholds turn numbers into readable categories.
# Conditional branches must check boundaries carefully.
# This example classifies familiar vital measurements.

# Store one small set of patient measurements.
temperature_c = 38.2
oxygen_percent = 93
systolic_bp = 146

# Classify temperature using ordered clinical thresholds.
if temperature_c >= 38.0:
    temperature_label = "fever range"
elif temperature_c >= 37.5:
    temperature_label = "elevated range"
else:
    temperature_label = "expected range"

# Classify oxygen saturation using clear cut points.
if oxygen_percent < 90:
    oxygen_label = "urgent low range"
elif oxygen_percent < 95:
    oxygen_label = "below expected range"
else:
    oxygen_label = "expected range"

# Classify systolic blood pressure using simple thresholds.
if systolic_bp >= 140:
    bp_label = "high range"
elif systolic_bp >= 120:
    bp_label = "elevated range"
else:
    bp_label = "expected range"

# Print a concise clinical summary for learners.
print("Threshold classification summary:")
print(f"Temperature: {temperature_c} C is {temperature_label}.")
print(f"Oxygen saturation: {oxygen_percent}% is {oxygen_label}.")
print(f"Systolic blood pressure: {systolic_bp} mmHg is {bp_label}.")

```

### **1.2. Threshold Based Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_01_02.jpg?v=1781140729" width="250">



>* Use thresholds to classify medical measurements
>* Apply predefined rules with Python conditionals

>* Order conditions to avoid overlapping categories
>* Handle cutoff values deliberately and clearly

>* Turn measurements into readable clinical categories
>* Use thresholds as decision support only



```
#@title Python Code - Threshold Based Classification

# Classify measurements with simple clinical thresholds.
# Use predefined cutoffs for consistent labels.
# This example is educational decision support.

# Store small sample readings in Celsius.
temperatures_celsius = [36.8, 37.5, 38.2, 39.1]

# Define a reusable temperature classifier.
def classify_temperature(temp_celsius):
    # Check the highest concern range first.
    if temp_celsius >= 38.0:
        return "fever range"

    # Check the mildly elevated range next.
    elif temp_celsius >= 37.2:
        return "elevated range"

    # Use the remaining range as expected.
    else:
        return "expected range"

# Store small sample fasting glucose readings.
glucose_mg_dl = [68, 91, 112, 131]

# Define a glucose screening classifier.
def classify_glucose(value_mg_dl):
    # Check below-range values first.
    if value_mg_dl < 70:
        return "low glucose"

    # Check typical fasting values next.
    elif value_mg_dl <= 99:
        return "typical fasting range"

    # Check elevated screening values next.
    elif value_mg_dl <= 125:
        return "elevated fasting range"

    # Classify values above this threshold.
    else:
        return "markedly high range"

# Print a short summary header.
print("Threshold based classification examples")

# Loop through temperatures and classify each reading.
for temp_celsius in temperatures_celsius:
    category = classify_temperature(temp_celsius)
    print(f"Temperature {temp_celsius} C: {category}")

# Loop through glucose values and classify each reading.
for value_mg_dl in glucose_mg_dl:
    category = classify_glucose(value_mg_dl)
    print(f"Glucose {value_mg_dl} mg/dL: {category}")

```

### **1.3. Threshold Based Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_01_03.jpg?v=1781140728" width="250">



>* Classify measurements using clinical cutoff values
>* Conditionals follow human-defined interpretation rules

>* Define exact cutoff behavior clearly
>* Use precise logic for consistent classifications

>* Interpret thresholds within clinical context
>* Encode clear rules, not universal truth



```
#@title Python Code - Threshold Based Classification

# Practice clinical threshold rules.
# Compare values against cutoffs.
# Print concise readable summaries.

# Store small example measurements.
measurements = [
    ("Patient A", 98.4),
    ("Patient B", 100.4),
    ("Patient C", 102.2),
]

# Define clinically inspired temperature thresholds.
elevated_threshold = 99.5
fever_threshold = 100.4

# Classify each temperature using ordered conditions.
for patient_name, temp_f in measurements:
    if temp_f >= fever_threshold:
        category = "fever range"

    elif temp_f >= elevated_threshold:
        category = "elevated temperature"

    else:
        category = "within expected range"

    # Print one clear clinical summary.
    print(patient_name, temp_f, "F:", category)

# Show exact boundary handling.
boundary_temp = fever_threshold

# Classify the boundary value explicitly.
if boundary_temp >= fever_threshold:
    boundary_category = "fever range"
else:
    boundary_category = "below fever threshold"

print("Boundary", boundary_temp, "F:", boundary_category)

```

## **2. Repeated Readings**

### **2.1. For Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_02_01.jpg?v=1781140715" width="250">



>* Loops review clinical readings one by one
>* Repeated code becomes shorter and clearer

>* Check each reading one at a time
>* Apply consistent rules for clear summaries

>* Connect individual readings to broader clinical patterns
>* Summarize repeated observations clearly and consistently



```
#@title Python Code - For Loops

# Practice for loops with clinical readings.
# Each reading is checked consistently.
# Output stays short and readable.

daily_temperatures_c = [36.8, 37.4, 38.2, 37.9, 36.9]

# Count fever readings during repeated review.
fever_count = 0

# Review each temperature one at a time.
for day_number, temperature_c in enumerate(daily_temperatures_c, start=1):

    # Classify each reading using one threshold.
    if temperature_c >= 38.0:
        status = "possible fever"
        fever_count = fever_count + 1

    # Otherwise mark readings below threshold.
    else:
        status = "below fever threshold"

    # Print one readable clinical line.
    print(f"Day {day_number}: {temperature_c} C, {status}.")

# Summarize the whole repeated-reading review.
total_readings = len(daily_temperatures_c)

# Show how many readings crossed threshold.
print(f"Summary: {fever_count} of {total_readings} readings suggested fever.")

```

### **2.2. Reading Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_02_02.jpg?v=1781140717" width="250">



>* Process repeated clinical measurements one by one
>* Apply consistent checks to every reading

>* Process each clinical reading one at a time
>* Compare values and report meaningful categories

>* Summarize patterns across repeated clinical readings
>* Turn measurements into clearer decision support



```
#@title Python Code - Reading Loops

# Practice looping through repeated clinical readings.
# Each value receives the same threshold logic.
# The summary counts readings needing attention.

# Store daily fasting glucose readings.
glucose_readings = [92, 108, 131, 84, 145, 99, 76]

# Prepare counters for the loop.
in_range_count = 0
high_count = 0
low_count = 0

# Review each glucose reading once.
for day_number, glucose in enumerate(glucose_readings, start=1):
    # Classify low readings first.
    if glucose < 80:
        category = "low"
        low_count = low_count + 1

    # Classify high readings next.
    elif glucose > 130:
        category = "above target"
        high_count = high_count + 1

    # Classify remaining readings as target.
    else:
        category = "within target"
        in_range_count = in_range_count + 1

    # Print one readable daily result.
    print(f"Day {day_number}: {glucose} mg/dL is {category}.")

# Count all reviewed readings safely.
total_readings = len(glucose_readings)

# Print a concise clinical summary.
print("Summary:")
print(f"Reviewed {total_readings} fasting glucose readings.")
print(f"Within target: {in_range_count}; high: {high_count}; low: {low_count}.")
print("Loop complete: each reading was checked consistently.")

```

### **2.3. While Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_02_03.jpg?v=1781140718" width="250">



>* Repeat readings until a condition changes
>* Stop when the clinical goal is met

>* Plan what changes each loop cycle
>* Move forward to avoid endless loops

>* Model ongoing monitoring with conditional checks
>* Stop when data, thresholds, or situations change



```
#@title Python Code - While Loops

# While loops support repeated clinical monitoring.
# Conditions decide whether monitoring should continue.
# Each cycle must move toward stopping.

# Store glucose readings collected during monitoring.
glucose_readings = [62, 68, 74, 83, 91]

# Start at the first available reading.
reading_index = 0

# Count reviewed readings for a summary.
reviewed_count = 0

# Continue while readings remain and values are low.
while reading_index < len(glucose_readings):
    current_glucose = glucose_readings[reading_index]

    # Stop when glucose reaches the safer range.
    if current_glucose >= 70:
        break

    # Report each low reading found.
    print("Low glucose reviewed:", current_glucose, "mg/dL")

    # Update counters so the loop advances.
    reviewed_count = reviewed_count + 1
    reading_index = reading_index + 1

# Check why the monitoring loop ended.
if reading_index < len(glucose_readings):
    safe_value = glucose_readings[reading_index]

    # Summarize the first safer glucose value.
    print("First safer value:", safe_value, "mg/dL")
else:
    safe_value = None

# Print a short clinical summary.
print("Low readings reviewed before stopping:", reviewed_count)

```

## **3. Logical Clinical Tests**

### **3.1. Combining Clinical Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_03_01.jpg?v=1781140721" width="250">



>* Combine observations to clarify clinical meaning
>* Use simple Python rules for readable summaries

>* Link values to interpret clinical meaning
>* Summarize repeated measures with combined signs

>* Clear conditions make summaries understandable
>* Accurate logic prevents misleading clinical messages



```
#@title Python Code - Combining Clinical Conditions

# Combine clinical findings into readable summaries.
# Use logical operators with clinical thresholds.
# Keep printed output short and clear.

# Define one small patient observation set.
patient_name = "Patient A"
temperature_c = 38.3
heart_rate_bpm = 112

# Add oxygen saturation and symptom information.
oxygen_percent = 94
confusion_present = False
fasting_glucose_mg_dl = 142

# Build clear boolean clinical conditions.
has_fever = temperature_c >= 38.0
has_fast_pulse = heart_rate_bpm > 100
has_low_oxygen = oxygen_percent < 95

# Combine conditions with logical operators.
possible_infection_pattern = has_fever and has_fast_pulse
respiratory_concern = has_low_oxygen or confusion_present
high_fasting_glucose = fasting_glucose_mg_dl >= 126

# Create a compact summary list.
summary = []

# Add infection pattern when both signs appear.
if possible_infection_pattern:
    summary.append("fever with fast pulse")
else:
    summary.append("no combined fever-pulse pattern")

# Add respiratory concern from either related condition.
if respiratory_concern:
    summary.append("respiratory or mental-status concern")
else:
    summary.append("oxygen and alertness reassuring")

# Add glucose interpretation using one threshold.
if high_fasting_glucose:
    summary.append("fasting glucose above diabetes threshold")
else:
    summary.append("fasting glucose below diabetes threshold")

# Print a readable clinical summary.
print(patient_name + " summary:")
print("- " + summary[0])
print("- " + summary[1])
print("- " + summary[2])

```

### **3.2. Combining Logical Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_03_02.jpg?v=1781140722" width="250">



>* Join comparisons for clinical decisions
>* Produce clear, readable patient summaries

>* Use and/or logic for clinical decisions
>* Match code exactly to clinical meaning

>* Explain clinical results in plain language
>* Group conditions for clear, trustworthy summaries



```
#@title Python Code - Combining Logical Conditions

# Combine logical tests for clinical summaries.
# Use thresholds, symptoms, and readable outputs.
# Keep the example small and deterministic.

# Store one patient's repeated measurements.
temperatures_c = [37.0, 38.3, 38.1]

# Store symptom information as booleans.
reports_chills = True
reports_breathlessness = False

# Store another key vital sign.
oxygen_saturation = 93

# Count elevated temperature readings.
elevated_count = 0
for reading in temperatures_c:
    if reading >= 38.0:
        elevated_count = elevated_count + 1

# Combine facts using and conditions.
fever_with_chills = elevated_count >= 2 and reports_chills

# Combine alternatives using or conditions.
urgent_flag = oxygen_saturation < 92 or reports_breathlessness

# Combine conditions for a summary decision.
needs_follow_up = fever_with_chills or urgent_flag

# Print concise clinical interpretation.
print("Elevated temperature readings:", elevated_count)
print("Fever with chills:", fever_with_chills)
print("Urgent respiratory flag:", urgent_flag)

# Choose a readable clinical message.
if needs_follow_up:
    summary = "Summary: follow-up recommended based on combined findings."
else:
    summary = "Summary: routine monitoring is reasonable today."

# Display the final clinical summary.
print(summary)

```

### **3.3. Not Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_C/image_03_03.jpg?v=1781140724" width="250">



>* Shows when clinical criteria are absent
>* Makes normal findings clear and readable

>* Not reverses the test, not facts
>* Tie results to specific criteria

>* Combine not with multiple clinical checks
>* Show unmet criteria for clearer summaries



```
#@title Python Code - Not Conditions

# Demonstrate not conditions in clinical summaries.
# Use thresholds tied to specific criteria.
# Keep output short and readable.

# Store one patient's simple measurements.
temperature_celsius = 37.4
oxygen_saturation = 96
symptom_present = False

# Define clinically named concern thresholds.
fever_threshold_celsius = 38.0
low_oxygen_cutoff = 92
urgent_symptom_required = True

# Build clear Boolean tests first.
meets_fever_criterion = temperature_celsius >= fever_threshold_celsius
meets_low_oxygen_criterion = oxygen_saturation < low_oxygen_cutoff
meets_symptom_criterion = symptom_present == urgent_symptom_required

# Combine tests into one urgent alert rule.
urgent_alert = meets_fever_criterion and meets_symptom_criterion
urgent_alert = urgent_alert or meets_low_oxygen_criterion

# Print the actual measurements first.
print("Clinical not-condition summary")
print(f"Temperature: {temperature_celsius} °C")
print(f"Oxygen saturation: {oxygen_saturation}%")

# Use not to state absent criteria clearly.
if not meets_fever_criterion:
    print("Does not meet the fever threshold.")

# Use not for low oxygen criterion.
if not meets_low_oxygen_criterion:
    print("Does not meet the low oxygen criterion.")

# Use not for combined urgent alert.
if not urgent_alert:
    print("No urgent alert from these defined rules.")
else:
    print("Urgent alert is triggered by these rules.")

```
