## **1. Medical Variables**

### **1.1. Variable Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_01_01.jpg?v=1781139716" width="250">



>* Variables name values for later use
>* Clear names give medical data meaning

>* Variables label values and can change
>* Clear names prevent clinical data confusion

>* Use clear, consistent medical variable names
>* Describe data types without exposing private details



```
#@title Python Code - Variable Basics

# Variables give medical values useful names.
# Clear names make health data readable.
# Reassignment updates changing clinical information.

# Store basic patient and measurement values.
patient_label = "Patient A"
heart_rate_bpm = 88
temperature_c = 37.2

# Store a true-or-false clinical status.
is_admitted = True
pain_score = 6

# Print the first recorded values.
print("Patient label:", patient_label)
print("Heart rate bpm:", heart_rate_bpm)
print("Temperature Celsius:", temperature_c)

# Print the clinical status value.
print("Currently admitted:", is_admitted)
print("Initial pain score:", pain_score)

# Reassign pain score after treatment.
pain_score = 3
print("Updated pain score:", pain_score)

# Inspect built-in data types.
print("Heart rate type:", type(heart_rate_bpm).__name__)
print("Patient label type:", type(patient_label).__name__)
print("Admission status type:", type(is_admitted).__name__)

```

### **1.2. Medical Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_01_02.jpg?v=1781139714" width="250">



>* Store clinical measurements in named variables
>* Use clear names to prevent confusion

>* Match numeric precision to clinical use
>* Track units carefully to prevent errors

>* Stored measurements support trends and decisions
>* Clear values improve accuracy and trust



```
#@title Python Code - Medical Measurements

# Store clinical measurements as named variables.
# Use numeric types matching measurement precision.
# Print concise results for quick inspection.

# Record common medical measurements clearly.
heart_rate_bpm = 82
body_temperature_c = 37.6
oxygen_saturation_percent = 94
patient_weight_kg = 68.5

# Record height using imperial units.
patient_height_inches = 66
height_meters = patient_height_inches * 0.0254
body_mass_index = patient_weight_kg / (height_meters ** 2)

# Inspect values and their Python types.
print("Heart rate:", heart_rate_bpm, type(heart_rate_bpm))
print("Temperature:", body_temperature_c, type(body_temperature_c))
print("Oxygen saturation:", oxygen_saturation_percent, type(oxygen_saturation_percent))
print("Weight in kg:", patient_weight_kg, type(patient_weight_kg))

# Convert BMI to a readable rounded value.
rounded_bmi = round(body_mass_index, 1)
print("Height in meters:", round(height_meters, 2))
print("Calculated BMI:", rounded_bmi)
print("Measurements stored successfully:", True)

```

## **2. Core Data Types**

### **2.1. Integer Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_02_01.jpg?v=1781139723" width="250">



>* Integers store whole-number clinical counts
>* Use them for clear simple calculations

>* Clinical observations often use whole-number units
>* Python treats integers as numeric data

>* Use integers when whole counts make sense
>* Treat numeric labels as text, not quantities



```
#@title Python Code - Integer Measurements

# Integers store whole clinical measurements.
# Types help separate quantities from labels.
# Conversions support simple medical calculations.

# Store whole-number clinical measurements.
patient_age_years = 42
heart_rate_bpm = 78
prior_admissions_count = 3
pain_score_scale = 7

# Store numeric-looking text as labels.
medical_record_number = "004215"
room_number_label = "12B"

# Inspect each value's Python type.
print("age type:", type(patient_age_years).__name__)
print("heart rate type:", type(heart_rate_bpm).__name__)
print("medical record type:", type(medical_record_number).__name__)

# Use integers in arithmetic calculations.
total_encounters = prior_admissions_count + 1
print("total encounters:", total_encounters)

# Convert text digits when arithmetic is meaningful.
steps_text = "1200"
steps_count = int(steps_text)
print("steps plus 300:", steps_count + 300)

# Keep identifiers as text, not quantities.
print("record label unchanged:", medical_record_number)
print("pain score is integer:", isinstance(pain_score_scale, int))

```

### **2.2. Floating Point Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_02_02.jpg?v=1781139719" width="250">



>* Floats store decimal clinical measurements
>* Use floats for measured, not counted, data

>* Floats handle tiny and huge decimals
>* Decimals may need careful rounding

>* Use floats for meaningful decimal measurements
>* Avoid floats for labels or statuses



```
#@title Python Code - Floating Point Values

# Floats store decimal medical measurement values.
# This example compares measurements and types.
# Run each line and read outputs.

# Store several measurements as float values.
temperature_c = 37.6
potassium_mmol_l = 4.2
newborn_weight_kg = 3.18
height_inches = 68.5

# Inspect the Python type of measurements.
print("Temperature type:", type(temperature_c).__name__)
print("Potassium type:", type(potassium_mmol_l).__name__)

# Use floats in a clinical threshold comparison.
has_low_grade_fever = temperature_c >= 37.5
print("Low-grade fever flag:", has_low_grade_fever)

# Convert an imperial height measurement to meters.
height_meters = height_inches * 0.0254
print("Height in meters:", round(height_meters, 2))

# Calculate body mass index using float arithmetic.
bmi_value = newborn_weight_kg / (height_meters ** 2)
print("Example BMI value:", round(bmi_value, 2))

# Convert text from a device into a float.
device_reading_text = "98.9"
device_temperature_f = float(device_reading_text)
print("Converted device type:", type(device_temperature_f).__name__)

```

### **2.3. String Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_02_03.jpg?v=1781139725" width="250">



>* Strings store clinical text and labels
>* Digit identifiers are not numeric measurements

>* Strings preserve exact clinical text formatting
>* Text categories are not numeric measurements

>* Keep text values consistent and carefully checked
>* Strings capture healthcare names, categories, and context



```
#@title Python Code - String Data

# This example explores clinical string values.
# Strings store text labels and identifiers.
# Python preserves exact characters in strings.

# Store medical information as string data.
patient_name = "Amina Patel"
medical_record_number = "007421"
clinic_location = "North Wing 3B"

# Store category-like clinical wording as strings.
triage_status = "urgent"
specimen_label = "LAB-00058-A"
allergy_note = "penicillin rash"

# Inspect the data type of text values.
print("MRN value:", medical_record_number)
print("MRN type:", type(medical_record_number).__name__)

# Compare string identifiers with numeric-looking values.
room_number = "012"
room_as_number = int(room_number)
print("Room string keeps zeros:", room_number)
print("Converted number drops zeros:", room_as_number)

# Combine strings to create a readable label.
summary_label = patient_name + " | " + triage_status
print("Visit label:", summary_label)

# Small spacing changes can affect equality.
status_one = "emergency"
status_two = "Emergency "
print("Exact text match:", status_one == status_two)

# Clean text before comparing categories.
clean_status = status_two.strip().lower()
print("Cleaned text match:", status_one == clean_status)

```

### **3.1. Inspect Value Types**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_03_01.jpg?v=1781139712" width="250">



>* Check how Python classifies each value

```
#@title Python Code - Inspect Value Types

# Inspect medical values before calculating.
# Python types guide safe operations.
# Small records make examples clear.

# Store example values from forms.
age_years = 54
weight_text = "72.5"
patient_id = "00731"
is_fasting = True

# Inspect each stored value type.
print("age_years type:", type(age_years))
print("weight_text type:", type(weight_text))
print("patient_id type:", type(patient_id))
print("is_fasting type:", type(is_fasting))

# Convert text weight for calculation.
weight_kg = float(weight_text)
weight_lb = weight_kg * 2.20462

# Show converted value and type.
print("weight_kg type:", type(weight_kg))
print("weight_lb rounded:", round(weight_lb, 1))
print("fasting value type:", type(is_fasting))

```

### **3.2. Rounding Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_03_02.jpg?v=1781139708" width="250">



>* Round medical calculations for clearer results
>* Choose precision appropriate to the context

>* Round displayed results, not working values
>* Keep precision for later clinical calculations

>* Use consistent rounding for clearer comparisons
>* Match precision to measurement and decision



```
#@title Python Code - Rounding Results

# Rounding makes clinical numbers easier to read.
# Keep precise values before final communication.
# Builtin functions support simple medical calculations.

# Store raw measurement values.
weight_kg = 72.4
height_m = 1.68

# Calculate BMI with full precision.
bmi_precise = weight_kg / (height_m ** 2)
bmi_display = round(bmi_precise, 1)

# Convert temperature and round result.
temp_f = 100.4
temp_c_precise = (temp_f - 32) * 5 / 9

# Round to one decimal place.
temp_c_display = round(temp_c_precise, 1)

# Convert height and round centimeters.
height_inches = 66.25
height_cm_precise = height_inches * 2.54
height_cm_display = round(height_cm_precise, 2)

# Print raw and rounded results.
print("Precise BMI:", bmi_precise)
print("Rounded BMI:", bmi_display)
print("Rounded temperature Celsius:", temp_c_display)
print("Rounded height centimeters:", height_cm_display)

```

### **3.3. Length Function**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_01/Lecture_B/image_03_03.jpg?v=1781139710" width="250">



>* Count items or characters in medical data
>* Check completeness and required data formats

>* Text length counts every visible character
>* Length checks reveal hidden data problems

>* Count observations before clinical calculations
>* Check completeness to avoid misleading results



```
#@title Python Code - Length Function

# Learn length with small medical examples.
# Use len to inspect data sizes.
# Counts support safer clinical calculations.

# Store text and list examples.
patient_code = "PT-00482"
status_label = "stable"
temperature_readings = [37.1, 37.4, 37.2]

# Measure characters inside text values.
code_length = len(patient_code)
status_length = len(status_label)
reading_count = len(temperature_readings)

# Calculate average only when data exists.
if reading_count > 0:
    average_temperature = sum(temperature_readings) / reading_count
else:
    average_temperature = None

# Show concise results for learners.
print("Patient code:", patient_code)
print("Code characters:", code_length)
print("Status label characters:", status_length)
print("Temperature reading count:", reading_count)

# Display average temperature safely.
print("Average temperature Celsius:", round(average_temperature, 1))

```
