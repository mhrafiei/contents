## **1. Readable Medical Summaries**

### **1.1. Formatted Tables**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_01_01.jpg?v=1781147099" width="250">



>* Structured tables make medical data easier to scan
>* String tools align fields and highlight key details

>* Use consistent dates, units, and labels
>* Standardize missing values for safer interpretation

>* Tables help different roles find key signals
>* Formatting supports comparison and responsible decisions



```
#@title Python Code - Formatted Tables

# Format rows into readable clinical tables.
# Built-in string methods control alignment.
# De-identification protects patient privacy.

# Create a tiny example audit dataset.
records = [
    ("Ava Patel", "2026-01-04", 148, 92, "clinic", "review booked"),
    ("Ben Morris", "2026-01-05", 126, 78, " ward ", "not recorded"),
    ("Cara Singh", "2026-01-06", 161, 95, "home", "call today"),
]

# Build concise de-identified patient labels.
def make_code(full_name, row_number):
    parts = full_name.strip().split()
    initials = "".join(part[0].upper() for part in parts)
    return f"{initials}-{row_number:02d}"

# Prepare fixed-width column headers.
headers = ["ID", "Date", "BP mmHg", "Setting", "Action"]
widths = [6, 12, 10, 10, 14]
header_row = " | ".join(text.ljust(width) for text, width in zip(headers, widths))
separator = "-+-".join("-" * width for width in widths)

# Convert records into formatted table rows.
rows = []
for number, record in enumerate(records, start=1):
    name, date, systolic, diastolic, setting, action = record
    patient_id = make_code(name, number)
    bp_text = f"{systolic}/{diastolic}"
    clean_setting = setting.strip().title()
    clean_action = action.strip().capitalize()
    values = [patient_id, date, bp_text, clean_setting, clean_action]
    rows.append(" | ".join(value.ljust(width) for value, width in zip(values, widths)))

# Print a compact readable report.
print("Readable blood pressure follow-up table")
print(header_row)
print(separator)
for row in rows:
    print(row)

# Highlight audit exceptions with narrative text.
high_risk = [row for row in records if row[2] >= 140 or row[3] >= 90]
summary = f"Audit note: {len(high_risk)} records need blood pressure follow-up."
print(summary)

```

### **1.2. Readable Rows**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_01_02.jpg?v=1781147101" width="250">



>* Consistent rows make records easy to scan
>* String tools reduce ambiguity and highlight patterns

>* Standardize names, dates, spacing, and units
>* Use string tools to join readable fields

>* Balance key details with easy scanning
>* Use strings for consistent, clear rows



```
#@title Python Code - Readable Rows

# Format clinical rows using built-in strings.
# Keep reports readable for quick review.
# De-identify identifiers before showing summaries.

# Create small example audit records.
records = [
    {"id": "MRN-48122", "date": "2026-04-03", "ward": " cardiology ",
     "bp": "168/96", "status": "follow-up booked"},
    {"id": "MRN-59203", "date": "2026-04-04", "ward": "RESPIRATORY",
     "bp": "142/88", "status": "pending review"},
    {"id": "MRN-61477", "date": "2026-04-04", "ward": "general medicine",
     "bp": "", "status": "documentation missing"},
]

# Convert identifiers into safer report labels.
def mask_identifier(raw_identifier):
    hidden_tail = raw_identifier[-2:]
    return "Patient-" + hidden_tail

# Standardize ward names and missing measurements.
def format_ward(raw_ward):
    cleaned_ward = raw_ward.strip()
    return cleaned_ward.title()

# Build one compact readable row.
def make_row(record):
    patient_label = mask_identifier(record["id"])
    ward_label = format_ward(record["ward"])
    bp_value = record["bp"] or "not recorded"
    status_text = record["status"].capitalize()
    return f"{record['date']} | {ward_label:<16} | {patient_label:<10} | BP: {bp_value:<12} | {status_text}"

# Print a clear heading and separator.
print("Readable medical summary rows")
print("Date       | Ward             | Patient    | Measurement     | Outcome")
print("-" * 78)

# Print each formatted row consistently.
for record in records:
    print(make_row(record))

# Add one short narrative summary.
missing_count = sum(1 for record in records if not record["bp"])
summary = f"Audit note: {missing_count} record needs blood pressure documentation."
print(summary)

```

### **1.3. Narrative Summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_01_03.jpg?v=1781147103" width="250">



>* Turn medical data into clear sentences
>* Use string tools for consistent formatting

>* Choose relevant fields for each report
>* Standardise messy text for clearer summaries

>* Match summaries to audience and purpose
>* Show missing data clearly and fairly



```
#@title Python Code - Narrative Summaries

# Build concise summaries from structured records.
# Clean strings before creating report sentences.
# De-identify examples before printing summaries.

# Create tiny example records for teaching.
records = [
    {
        "id": "MRN-88421",
        "initial": "A",
        "age": 67,

        "diagnosis": " pneumonia ",
        "stay_days": 3,
        "bp": "138/82",
        "follow_up": "yes",
    },

    {
        "id": "MRN-99217",
        "initial": "B",
        "age": 54,
        "diagnosis": "DIABETES REVIEW",

        "stay_days": 0,
        "bp": "151/94",
        "follow_up": "",
    },
]

# Define helper functions for consistent wording.
def clean_condition(text):
    cleaned = text.strip().lower()
    return cleaned.capitalize()

# Convert missing follow-up values safely.
def follow_up_phrase(value):
    cleaned = value.strip().lower()
    if cleaned == "yes":
        return "follow-up was arranged"
    return "follow-up was not recorded"

# De-identify identifiers before reporting.
def anonymise(record_number):
    suffix = record_number[-2:]
    return "Patient " + suffix

# Build one narrative sentence per record.
def make_summary(record):
    patient = anonymise(record["id"])
    condition = clean_condition(record["diagnosis"])
    follow_up = follow_up_phrase(record["follow_up"])

    sentence = (
        f"{patient}, aged {record['age']}, was reviewed for "
        f"{condition} with latest blood pressure {record['bp']} mmHg; "
        f"length of stay was {record['stay_days']} days, and {follow_up}."
    )
    return sentence

# Print readable narrative summaries only.
print("Narrative medical summaries:")
for record in records:
    print("- " + make_summary(record))

```

## **2. Sorting Clinical Records**

### **2.1. Sorting Patient Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_02_01.jpg?v=1781147095" width="250">



>* Arrange records by clinically useful fields
>* Reveal patterns without changing underlying data

>* Match sort order to clinical purpose
>* Layer sorting by location, time, priority

>* Check formats, numbers, identifiers, and missing values
>* Ensure sorting supports safe clinical decisions



```
#@title Python Code - Sorting Patient Records

# Sort records to highlight clinical priorities.
# Small examples keep reports readable.
# Deidentified labels protect patient privacy.

# Create a tiny clinical record list.
records = [
    {"label": "P001", "ward": "A", "days": 5, "risk": 8.4},
    {"label": "P002", "ward": "B", "days": 2, "risk": 4.1},

    {"label": "P003", "ward": "A", "days": 12, "risk": 9.2},
    {"label": "P004", "ward": "B", "days": 8, "risk": 7.5},
    {"label": "P005", "ward": "A", "days": 3, "risk": 5.6},
]

# Confirm the example remains small.
assert len(records) <= 6, "Too many demo records"

# Sort longest admissions to the top.
by_stay = sorted(
    records,
    key=lambda row: row["days"],
    reverse=True,
)

# Sort by ward, then risk.
by_ward_risk = sorted(
    records,
    key=lambda row: (row["ward"], -row["risk"]),
)

# Filter possible audit exceptions.
exceptions = [
    row for row in by_stay
    if row["days"] >= 7 or row["risk"] >= 8
]

# Print a compact deidentified report.
print("Longest stays, deidentified:")
for row in by_stay[:3]:
    text = f"{row['label']} ward {row['ward']}: {row['days']} days"
    print(text)

# Show ward ordering and risk priority.
print("\nWard order, highest risk first:")
for row in by_ward_risk[:3]:
    text = f"{row['ward']} {row['label']}: risk {row['risk']}"
    print(text)

# Count records needing review.
summary = f"\nAudit exceptions needing review: {len(exceptions)}"
print(summary)

```

### **2.2. Audit Exceptions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_02_02.jpg?v=1781147093" width="250">



>* Exceptions flag records needing closer review
>* Sorting makes unusual clinical patterns visible

>* Sort records to reveal cases needing action
>* Interpret exceptions using clinical and data context

>* Explain sorting logic and urgency clearly
>* Treat exceptions as actionable review signals



```
#@title Python Code - Audit Exceptions

# Sort records to reveal audit exceptions.
# Small examples protect patient privacy.
# Built-in tools make logic transparent.

# Create tiny de-identified clinical records.
records = [
    {"id": "P001", "ward": "A", "days": 3, "reconciled": True},
    {"id": "P002", "ward": "B", "days": 1, "reconciled": False},

    {"id": "P003", "ward": "A", "days": 5, "reconciled": False},
    {"id": "P004", "ward": "C", "days": 2, "reconciled": True},
    {"id": "P005", "ward": "B", "days": 4, "reconciled": False},
]

# Define the audit rule clearly.
target_days = 2

# Filter records that breach the standard.
exceptions = [
    rec for rec in records
    if rec["days"] > target_days and not rec["reconciled"]
]

# Sort most overdue records first.
exceptions_sorted = sorted(
    exceptions,
    key=lambda rec: rec["days"],
    reverse=True,
)

# Print a short readable audit table.
print("Medicines reconciliation audit exceptions")
print("Rule: not reconciled after more than 2 days")
print("ID    Ward    Days since admission")

# Display only highlighted exception records.
for rec in exceptions_sorted:
    row = f"{rec['id']:<5} {rec['ward']:<7} {rec['days']:>3}"
    print(row)

# Summarise the review workload.
summary = f"Review needed for {len(exceptions_sorted)} patient records."
print(summary)

```

### **2.3. Filtering Audit Exceptions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_02_03.jpg?v=1781147097" width="250">



>* Narrow records to possible review cases
>* Preserve context for professional judgement

>* Define exceptions as clear record conditions
>* Include concerns while excluding irrelevant records

>* Present filtered records as review candidates
>* Sort responsibly to support fair action



```
#@title Python Code - Filtering Audit Exceptions

# Filter audit exceptions from clinical records.
# Sort review cases by clinical priority.
# Deidentify records before reporting findings.

# Create tiny example clinic audit records.
records = [
    {"patient": "Amina Patel", "clinic": "Diabetes", "hba1c_days": 310, "allergy": "Penicillin"},
    {"patient": "Ben Wright", "clinic": "Diabetes", "hba1c_days": 120, "allergy": ""},
    {"patient": "Chen Li", "clinic": "Cardiology", "hba1c_days": None, "allergy": "NKDA"},
    {"patient": "Dina Evans", "clinic": "Diabetes", "hba1c_days": 455, "allergy": ""},
]

# Define audit rules as readable variables.
review_days_limit = 180
missing_text = ""

# Select records meeting exception criteria.
exceptions = []
for record in records:
    overdue = record["hba1c_days"] is not None and record["hba1c_days"] > review_days_limit
    missing_allergy = record["allergy"] == missing_text
    if record["clinic"] == "Diabetes" and (overdue or missing_allergy):
        exceptions.append(record)

# Sort exceptions so longest overdue appears first.
exceptions.sort(key=lambda item: item["hba1c_days"] or 0, reverse=True)

# Print a short deidentified audit report.
print("Records meeting audit review criteria:")
print("ID     Clinic      HbA1c days  Allergy note")
for number, record in enumerate(exceptions, start=1):
    safe_id = f"P{number:03d}"
    allergy_note = record["allergy"] or "MISSING"
    print(f"{safe_id:<6} {record['clinic']:<10} {record['hba1c_days']:<10} {allergy_note}")

# Summarise interpretation without confirming wrongdoing.
print(f"Total review records: {len(exceptions)}")

```

## **3. De Identification**

### **3.1. Remove Patient Identifiers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_03_01.jpg?v=1781147087" width="250">



>* Remove details that directly identify patients
>* Keep only data needed for reporting

>* Check every field for identifying details
>* Generalize notes and narratives before reporting

>* Combined details can reveal patient identities
>* Share only needed, carefully checked data



```
#@title Python Code - Remove Patient Identifiers

# Remove direct identifiers before reporting.
# Keep only fields needed for summaries.
# Review narrative text for hidden identifiers.

# Build a tiny example clinic dataset.
records = [
    {
        "name": "Maria Khan",
        "mrn": "MRN-1001",
        "phone": "555-0191",

        "email": "maria.k@example.org",
        "clinic": "Cardiology",
        "wait_minutes": 20,
        "comment": "Asked for results by email",
    },
    {
        "name": "James Liu",
        "mrn": "MRN-1002",
        "phone": "555-0177",

        "email": "jliu@example.org",
        "clinic": "Respiratory",
        "wait_minutes": 46,
        "comment": "Mayor's son requested side room",
    },
]

# List columns that directly identify patients.
identifier_fields = ["name", "mrn", "phone", "email"]

# List columns useful for the audit report.
report_fields = ["clinic", "wait_minutes", "comment"]

# Create safe records without direct identifiers.
deidentified = []
for record in records:
    safe_record = {}
    for field in report_fields:
        safe_record[field] = record[field]
    deidentified.append(safe_record)

# Replace risky narrative details with safer wording.
risky_phrase = "Mayor's son requested side room"
safe_phrase = "Family requested side room"
for record in deidentified:
    record["comment"] = record["comment"].replace(risky_phrase, safe_phrase)

# Print a compact audit-ready report.
print("Original fields:", sorted(records[0].keys()))
print("Removed identifiers:", identifier_fields)
print("Reporting fields:", sorted(deidentified[0].keys()))
print("De-identified rows:")
for index, record in enumerate(deidentified, start=1):
    print(index, record)

# Confirm no direct identifier fields remain.
remaining = set(deidentified[0].keys()) & set(identifier_fields)
print("Direct identifiers remaining:", len(remaining))

```

### **3.2. Safe Data Masking**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_03_02.jpg?v=1781147091" width="250">



>* Replace sensitive details with broader categories
>* Keep useful patterns while reducing identification risk

>* Match masking to the report purpose
>* Check notes and fields for hidden identifiers

>* Avoid masking that still permits re-identification
>* Generalize rare or unnecessary details proportionally



```
#@title Python Code - Safe Data Masking

# Demonstrate masking before sharing reports.
# Keep useful clinical summary patterns.
# Remove unnecessary identifying detail.

# Import pandas for small tabular examples.
import pandas as pd

# Create a tiny teaching dataset.
records = [
    {"name": "Amira Patel", "mrn": "A12345", "age": 22},
    {"name": "Ben Jones", "mrn": "B67890", "age": 87},
    {"name": "Chloe Smith", "mrn": "C24680", "age": 41},
]

# Add location and timing details.
records[0]["postcode"] = "SW1A 1AA"
records[1]["postcode"] = "YO10 5DD"
records[2]["postcode"] = "AB12 3CD"
records[0]["visit_date"] = "2026-02-03"

# Complete the remaining visit dates.
records[1]["visit_date"] = "2026-02-06"
records[2]["visit_date"] = "2026-02-20"
records[0]["note"] = "Vaccination review completed."
records[1]["note"] = "Lives beside named rural school."

# Add a safer generic note.
records[2]["note"] = "Medication review completed."
data = pd.DataFrame(records)
assert len(data) == 3

# Convert exact ages into bands.
def age_band(age):
    if age < 25:
        return "18-24"
    if age < 65:
        return "25-64"
    return "65+"

# Keep district only from postcode.
def postcode_area(postcode):
    return postcode.split()[0]

# Replace notes that may contain context clues.
def safe_note(text):
    if "named" in text or "school" in text:
        return "Potential identifier removed."
    return text

# Build a masked reporting table.
masked = pd.DataFrame()
masked["study_id"] = [f"patient_{index+1}" for index in range(len(data))]
masked["age_band"] = data["age"].apply(age_band)
masked["area"] = data["postcode"].apply(postcode_area)

# Show timing by month only.
visit_months = pd.to_datetime(data["visit_date"]).dt.to_period("M")
masked["visit_month"] = visit_months.astype(str)
masked["note"] = data["note"].apply(safe_note)

# Print only the masked report.
print("Masked audit report:")
print(masked.to_string(index=False))
print("\nOriginal names and MRNs are not reported.")

```

### **3.3. Limited Data Reporting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_B/image_03_03.jpg?v=1781147089" width="250">



>* Share only data needed for the question
>* Avoid details that could identify people

>* Combined details can indirectly identify patients
>* Use broader categories to protect privacy

>* Review details before sharing reports
>* Combine small groups to protect privacy



```
#@title Python Code - Limited Data Reporting

# Demonstrate limited reporting for clinical audit.
# Keep only fields needed for summaries.
# Replace direct identifiers with broader categories.

# Import pandas for small table summaries.
import pandas as pd

# Create a tiny example appointment dataset.
records = [
    {"name": "Amina Patel", "dob": "1974-03-12", "town": "Riverton", "clinic": "North Diabetes", "appt_date": "2024-01-08", "age": 49, "referral": "routine", "followed_up": True},
    {"name": "Ben Jones", "dob": "1939-11-02", "town": "Riverton", "clinic": "North Diabetes", "appt_date": "2024-01-15", "age": 84, "referral": "urgent", "followed_up": False},
    {"name": "Carla Gomez", "dob": "1965-06-20", "town": "Lakehill", "clinic": "East Diabetes", "appt_date": "2024-02-05", "age": 58, "referral": "routine", "followed_up": True},
    {"name": "Dev Singh", "dob": "1988-09-09", "town": "Pineford", "clinic": "Rural Diabetes", "appt_date": "2024-02-10", "age": 35, "referral": "urgent", "followed_up": False},
]

# Convert records into a small DataFrame.
df = pd.DataFrame(records)

# Convert appointment dates for month reporting.
df["appt_date"] = pd.to_datetime(df["appt_date"])

# Define age bands instead of exact ages.
df["age_band"] = pd.cut(
    df["age"], bins=[0, 39, 64, 120], labels=["0-39", "40-64", "65+"]
)

# Report month instead of full appointment date.
df["appt_month"] = df["appt_date"].dt.to_period("M").astype(str)

# Keep only fields needed for audit reporting.
limited = df[["age_band", "clinic", "appt_month", "referral", "followed_up"]]

# Summarize follow-up counts by limited fields.
summary = limited.groupby(
    ["age_band", "referral", "followed_up"], observed=True
).size().reset_index(name="count")

# Suppress very small cell counts before release.
summary["reported_count"] = summary["count"].apply(lambda value: "<2" if value < 2 else str(value))

# Display only the limited release table.
print("Limited report table, identifiers removed:")

# Show compact output for teaching.
print(summary[["age_band", "referral", "followed_up", "reported_count"]].to_string(index=False))

# Add a brief narrative interpretation.
message = "Names, birthdates, towns, and exact dates were not reported."
print(message)

```
