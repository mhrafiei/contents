## **1. Basic Medical Statistics**

### **1.1. Core Summary Measures**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_01_01.jpg?v=1781146104" width="250">



>* Summaries turn records into clear clinical information
>* Counts, totals, averages, minima, and maxima

>* Know what each record represents
>* Check missing, invalid, or unusual values

>* Summary measures reveal hidden clinical patterns
>* They guide questions, not clinical judgment



```
#@title Python Code - Core Summary Measures

# Summaries turn records into clinical measures.
# Small datasets make calculations easy to inspect.
# Built-in functions support accurate quick summaries.

# No package installation is needed.

# Each dictionary represents one patient admission.
admissions = [
    {"ward": "ED", "age": 67, "stay_hours": 10, "systolic_bp": 148},
    {"ward": "ED", "age": 45, "stay_hours": 6, "systolic_bp": 132},
    {"ward": "Medical", "age": 72, "stay_hours": 48, "systolic_bp": 156},
    {"ward": "Medical", "age": 58, "stay_hours": 36, "systolic_bp": 140},
    {"ward": "Surgical", "age": 39, "stay_hours": 30, "systolic_bp": 125},
    {"ward": "Surgical", "age": 51, "stay_hours": 42, "systolic_bp": 138},
]

# Count records before calculating other statistics.
record_count = len(admissions)

# Extract numeric fields into simple lists.
ages = [record["age"] for record in admissions]
stay_hours = [record["stay_hours"] for record in admissions]
systolic_values = [record["systolic_bp"] for record in admissions]

# Check that every list matches record count.
valid_lengths = all(len(values) == record_count for values in [ages, stay_hours, systolic_values])

# Calculate totals and averages using built-ins.
total_stay = sum(stay_hours)
average_age = sum(ages) / record_count
average_systolic = sum(systolic_values) / record_count

# Find minimum and maximum clinical observations.
shortest_stay = min(stay_hours)
longest_stay = max(stay_hours)
highest_systolic = max(systolic_values)

# Print compact summary lines for learners.
print("Records checked:", record_count, "valid lengths:", valid_lengths)
print("Total stay hours:", total_stay)
print("Average age:", round(average_age, 1), "years")
print("Average systolic BP:", round(average_systolic, 1), "mmHg")
print("Stay range:", shortest_stay, "to", longest_stay, "hours")
print("Highest systolic BP:", highest_systolic, "mmHg")

```

### **1.3. Core Summary Measures**

```
#@title Python Code - Core Summary Measures

# Summaries turn records into practical clinical information.
# Small datasets make each calculation easy to inspect.
# Built-in functions support accurate beginner friendly analysis.

# Create a tiny ward dataset.
records = [
    {"patient": "P001", "ward": "Medical", "stay_days": 3, "wait_min": 42},
    {"patient": "P002", "ward": "Surgical", "stay_days": 5, "wait_min": 25},

    {"patient": "P003", "ward": "Medical", "stay_days": 2, "wait_min": 35},
    {"patient": "P004", "ward": "Pediatric", "stay_days": 1, "wait_min": 18},

    {"patient": "P005", "ward": "Surgical", "stay_days": 4, "wait_min": 50},
]

# Count records, representing patients here.
patient_count = len(records)

# Extract numeric fields for summary measures.
stay_days = [record["stay_days"] for record in records]
wait_minutes = [record["wait_min"] for record in records]

# Calculate totals and averages safely.
total_stay_days = sum(stay_days)
average_stay_days = total_stay_days / patient_count
average_wait = sum(wait_minutes) / patient_count

# Calculate minimum and maximum values.
shortest_stay = min(stay_days)
longest_stay = max(stay_days)
maximum_wait = max(wait_minutes)

# Display concise summary results.
print("Basic ward summary")
print(f"Patient count: {patient_count}")
print(f"Total stay days: {total_stay_days}")
print(f"Average stay days: {average_stay_days:.1f}")
print(f"Shortest and longest stays: {shortest_stay}, {longest_stay}")
print(f"Average wait minutes: {average_wait:.1f}")
print(f"Maximum wait minutes: {maximum_wait}")

```

## **2. Grouping Clinical Records**

### **2.1. Grouping by Ward**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_02_01.jpg?v=1781146095" width="250">



>* Arrange patient records by hospital ward
>* Compare ward activity, risks, and needs

>* Standardize ward names before grouping records
>* Match grouping detail to the clinical question

>* Ward groups enable useful summary measures
>* Interpret ward patterns in clinical context



```
#@title Python Code - Grouping by Ward

# Group ward labels before summarizing records.
# Small dictionaries can represent patient rows.
# Standard library counters support clear grouping.

from collections import defaultdict

# Create a tiny hospital record list.
records = [
    {"patient": "P001", "ward": "ICU", "stay_days": 5, "risk": "High"},
    {"patient": "P002", "ward": "Medical", "stay_days": 3, "risk": "Low"},

    {"patient": "P003", "ward": "ICU", "stay_days": 8, "risk": "High"},
    {"patient": "P004", "ward": "Surgical", "stay_days": 2, "risk": "Moderate"},

    {"patient": "P005", "ward": "Medical", "stay_days": 6, "risk": "High"},
    {"patient": "P006", "ward": "Surgical", "stay_days": 4, "risk": "Low"},
]

# Group records under each ward name.
ward_groups = defaultdict(list)
for record in records:
    ward_groups[record["ward"]].append(record)

# Print compact summaries for each ward.
print("Ward summaries from grouped records")
for ward, patients in sorted(ward_groups.items()):
    stays = [patient["stay_days"] for patient in patients]

    high_count = sum(patient["risk"] == "High" for patient in patients)
    average_stay = sum(stays) / len(stays)

    summary = (
        f"{ward}: {len(patients)} patients, "
        f"avg stay {average_stay:.1f} days, "
        f"high risk {high_count}"
    )

    print(summary)

```

### **2.2. Grouping by Ward**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_02_02.jpg?v=1781146093" width="250">



>* Organize records by ward to reveal patterns
>* Preserve clinical context in manageable summaries

>* Standardize ward labels before summarizing records
>* Choose grouping rules matching the clinical question

>* Compare ward workload, safety, and resource needs
>* Interpret differences using clinical context



```
#@title Python Code - Grouping by Ward

# Group ward records into simple summaries.
# Small examples protect patient privacy.
# Built-in tools keep calculations clear.

from collections import defaultdict

# Create tiny clinical records.
records = [
    {"ward": "Emergency", "patients": 18, "falls": 1},
    {"ward": "ICU", "patients": 9, "falls": 0},

    {"ward": "Emergency", "patients": 22, "falls": 2},
    {"ward": "Surgical", "patients": 14, "falls": 1},
    {"ward": "ICU", "patients": 11, "falls": 1},

    {"ward": "Surgical", "patients": 16, "falls": 0},
]

# Prepare grouped totals by ward.
summary = defaultdict(lambda: {"days": 0, "patients": 0, "falls": 0})

# Add each record to its ward group.
for record in records:
    ward = record["ward"]
    summary[ward]["days"] += 1
    summary[ward]["patients"] += record["patients"]

    summary[ward]["falls"] += record["falls"]

# Print one compact summary line per ward.
print("Ward summary from daily records")
for ward in sorted(summary):
    data = summary[ward]
    average = data["patients"] / data["days"]

    message = f"{ward}: {data['patients']} patients, avg {average:.1f}, falls {data['falls']}"
    print(message)

# Show why grouping supports interpretation.
highest = max(summary, key=lambda ward: summary[ward]["patients"])
print(f"Highest total patient load: {highest}")

```

### **2.3. Grouping by Ward**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_02_03.jpg?v=1781146097" width="250">



>* Group patient records by ward location
>* Compare workload, risk, and care patterns

>* Standardize ward names before grouping records
>* Use clean groups for accurate summaries

>* Ward summaries reveal workload and resource needs
>* Interpret ward differences as inquiry, not judgment



```
#@title Python Code - Grouping by Ward

# Grouping ward records supports clinical summaries.
# Small lists can represent patient encounters.
# Built-in tools calculate ward-level measures.

from collections import defaultdict
from statistics import mean

# Create a tiny clinical dataset.
records = [
    {"patient_id": "P001", "ward": "Medical", "stay_days": 3, "risk": "high"},
    {"patient_id": "P002", "ward": "Surgical", "stay_days": 5, "risk": "low"},
    {"patient_id": "P003", "ward": "Medical", "stay_days": 7, "risk": "high"},
    {"patient_id": "P004", "ward": "ICU", "stay_days": 2, "risk": "high"},

    {"patient_id": "P005", "ward": "Surgical", "stay_days": 4, "risk": "medium"},
    {"patient_id": "P006", "ward": "ICU", "stay_days": 6, "risk": "medium"},
]

# Confirm records exist before grouping.
if len(records) == 0:
    raise ValueError("Records list must not be empty")

# Group each record under its ward.
ward_groups = defaultdict(list)
for record in records:
    ward_groups[record["ward"]].append(record)

# Print a compact ward-level summary.
print("Ward summary: count, total days, average, minimum, maximum")
for ward, patients in sorted(ward_groups.items()):
    stays = [patient["stay_days"] for patient in patients]
    high_count = sum(patient["risk"] == "high" for patient in patients)

    summary = (
        f"{ward}: n={len(patients)}, total={sum(stays)}, "
        f"avg={mean(stays):.1f}, min={min(stays)}, max={max(stays)}, "
        f"high-risk={high_count}"
    )
    print(summary)

print("Grouping by ward turns individual records into location summaries.")

```

## **3. Helpful Summary Modules**

### **3.1. Standard Summary Tools**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_03_01.jpg?v=1781146101" width="250">



>* Summarize medical records with reliable standard tools
>* Reduce errors and focus on meaningful questions

>* Consistent tools simplify varied healthcare summaries
>* Clear workflows reduce mistakes and support decisions

>* Prepare records before running summary calculations
>* Combine efficient tools with clinical judgment



```
#@title Python Code - Standard Summary Tools

# Summaries help clinicians review small record sets.
# Standard modules reduce repetitive calculation steps.
# This example uses safe built-in tools.

from collections import Counter, defaultdict
from statistics import mean
from datetime import date

# Define small medical visit records.
records = [
    {"ward": "Medical", "risk": "High", "age": 72, "bed_days": 5, "temp_c": 38.2, "admit": date(2024, 1, 3)},
    {"ward": "Surgical", "risk": "Low", "age": 45, "bed_days": 2, "temp_c": 36.9, "admit": date(2024, 1, 5)},
    {"ward": "Medical", "risk": "Medium", "age": 61, "bed_days": 4, "temp_c": 37.5, "admit": date(2024, 1, 4)},
    {"ward": "ICU", "risk": "High", "age": 80, "bed_days": 8, "temp_c": 39.1, "admit": date(2024, 1, 2)},
    {"ward": "Surgical", "risk": "Medium", "age": 54, "bed_days": 3, "temp_c": 37.2, "admit": date(2024, 1, 6)},
]

# Validate the dataset before summarizing.
if not records:
    raise ValueError("The record list must not be empty.")

# Extract numeric fields with list comprehensions.
ages = [record["age"] for record in records]
bed_days = [record["bed_days"] for record in records]
temperatures = [record["temp_c"] for record in records]
admit_dates = [record["admit"] for record in records]

# Count visits by ward and risk.
ward_counts = Counter(record["ward"] for record in records)
risk_counts = Counter(record["risk"] for record in records)

# Total bed days within each ward.
ward_bed_days = defaultdict(int)
for record in records:
    ward_bed_days[record["ward"]] += record["bed_days"]

# Print compact overall summaries.
print(f"Total visits: {len(records)}")
print(f"Total bed days: {sum(bed_days)}")
print(f"Average age: {mean(ages):.1f} years")
print(f"Temperature range: {min(temperatures)} to {max(temperatures)} C")

# Print date and grouped summaries.
print(f"Admission dates: {min(admit_dates)} to {max(admit_dates)}")
print(f"Visits by ward: {dict(ward_counts)}")
print(f"Visits by risk: {dict(risk_counts)}")
print(f"Bed days by ward: {dict(ward_bed_days)}")

```

### **3.2. Counter summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_03_02.jpg?v=1781146102" width="250">



>* Counter automatically counts repeated medical categories
>* Compact frequencies turn records into clear summaries

>* Count repeated clinical categories reliably
>* Spot unexpected labels needing review

>* Check category counts against total records
>* Use frequencies to guide further analysis



```
#@title Python Code - Counter summaries

# Counter quickly summarizes repeated medical categories.
# This example uses tiny ward records.
# Counts help check data distributions.

# Import Counter from the standard library.
from collections import Counter

# Create a small list of patient ward labels.
ward_labels = [
    "ICU", "Medical", "Medical",
    "Surgical", "ICU", "Medical",
    "Rehab", "Surgical", "Medical",
]

# Count how often each ward appears.
ward_counts = Counter(ward_labels)

# Count the total number of records.
total_records = sum(ward_counts.values())

# Display a compact frequency summary.
print("Ward frequency summary:")

# Show categories ordered by frequency.
for ward, count in ward_counts.most_common():
    percent = count / total_records * 100
    print(f"{ward}: {count} records ({percent:.1f}%)")

# Confirm that every record was counted.
print(f"Total counted records: {total_records}")

# Identify the most common ward category.
most_common_ward, most_common_count = ward_counts.most_common(1)[0]

# Print the leading category clearly.
print(f"Most common ward: {most_common_ward} ({most_common_count})")

```

### **3.3. Default Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_04/Lecture_A/image_03_03.jpg?v=1781146099" width="250">



>* Automatically starts new category summaries
>* Reduces repetitive checks and summary errors

>* Accumulate records or totals by category
>* Simplify grouping for consistent summaries

>* Organize records before calculating summaries
>* Reduce key errors and repetitive bookkeeping



```
#@title Python Code - Default Dictionaries

# Default dictionaries simplify grouped medical summaries.
# Each missing category receives a starting value.
# This example summarizes short patient records.

from collections import defaultdict

# Store a tiny list of ward records.
records = [
    {"ward": "Cardiology", "days": 4},
    {"ward": "Surgery", "days": 2},
    {"ward": "Cardiology", "days": 6},
    {"ward": "Medicine", "days": 3},
    {"ward": "Surgery", "days": 5},
    {"ward": "Medicine", "days": 4},
]

# Create grouped lists without checking keys.
days_by_ward = defaultdict(list)

# Add each length of stay to its ward.
for record in records:
    ward = record["ward"]
    days = record["days"]
    days_by_ward[ward].append(days)

# Print one compact summary per ward.
for ward, days_list in sorted(days_by_ward.items()):
    count = len(days_list)
    total = sum(days_list)
    average = total / count
    low = min(days_list)
    high = max(days_list)
    print(f"{ward}: n={count}, total={total}, avg={average:.1f}, min={low}, max={high}")

```
