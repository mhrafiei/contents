## **1. Input Validation Checks**

### **1.1. Missing Input Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_01_01.jpg?v=1781144906" width="250">



>* Confirm all required clinical data is present
>* Prevent misleading results from incomplete inputs

>* Separate required, optional, and assumed inputs
>* Clarify blanks, unknowns, and not-applicable values

>* Explain missing data and next steps
>* Warn, stop, or continue with limits



```
#@title Python Code - Missing Input Checks

# Demonstrate missing clinical input checks.
# Required fields must be present.
# Warnings explain calculation limitations.

# Create a tiny patient record.
patient = {
    "age_years": 67,
    "weight_kg": None,

    "height_cm": 170,
    "creatinine_mg_dl": "unknown",
}

# Define placeholders treated as missing.
missing_markers = {None, "", "unknown", "not recorded"}
required_fields = ["age_years", "weight_kg", "height_cm"]
optional_fields = ["creatinine_mg_dl"]

# Collect missing required inputs.
missing_required = []
for field in required_fields:
    value = patient.get(field)

    if value in missing_markers:
        missing_required.append(field)

# Collect missing optional inputs.
missing_optional = []
for field in optional_fields:
    value = patient.get(field)

    if value in missing_markers:
        missing_optional.append(field)

# Report validation findings clearly.
print("Missing input validation summary")
print("Required missing:", missing_required or "none")
print("Optional missing:", missing_optional or "none")

# Stop BMI if essential inputs are absent.
if missing_required:
    print("BMI not calculated: required data are missing.")

else:
    height_m = patient["height_cm"] / 100
    bmi = patient["weight_kg"] / (height_m ** 2)

    print("BMI:", round(bmi, 1), "kg/m^2")

# Explain optional missing information.
if missing_optional:
    print("Warning: kidney function interpretation is incomplete.")

print("Clinical calculations need complete required inputs.")

```

### **1.2. Range Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_01_02.jpg?v=1781144908" width="250">



>* Check values are clinically plausible
>* Block questionable inputs before calculations

>* Ranges depend on patient and clinical context
>* Check units match calculation assumptions

>* Separate impossible, unlikely, and extreme values
>* Use hard stops and verification warnings



```
#@title Python Code - Range Checks

# Range checks protect clinical calculations.
# Hard limits block impossible values.
# Soft limits warn about extremes.

# Define adult validation ranges.
ranges = {
    "age_years": {"hard": (0, 120), "soft": (18, 100)},
    "height_cm": {"hard": (80, 250), "soft": (140, 210)},
    "heart_rate_bpm": {"hard": (20, 250), "soft": (45, 130)},
}

# Create small example patient records.
patients = [
    {"id": "A", "age_years": 45, "height_cm": 175, "heart_rate_bpm": 75},
    {"id": "B", "age_years": -3, "height_cm": 168, "heart_rate_bpm": 80},
    {"id": "C", "age_years": 70, "height_cm": 1.75, "heart_rate_bpm": 750},
    {"id": "D", "age_years": 92, "height_cm": 188, "heart_rate_bpm": 142},
]

# Check one value against hard and soft limits.
def check_range(name, value):
    limits = ranges[name]
    hard_low, hard_high = limits["hard"]
    soft_low, soft_high = limits["soft"]

    # Reject missing or nonnumeric values.
    if value is None or not isinstance(value, (int, float)):
        return "BLOCK", f"{name} is missing or nonnumeric"

    # Block impossible values outside hard limits.
    if value < hard_low or value > hard_high:
        return "BLOCK", f"{name}={value} outside hard range"

    # Warn for unusual values outside soft limits.
    if value < soft_low or value > soft_high:
        return "WARN", f"{name}={value} is clinically unusual"

    # Accept values inside both ranges.
    return "OK", f"{name}={value} accepted"

# Validate each patient record safely.
for patient in patients:
    messages = []
    blocked = False

    # Apply each range rule consistently.
    for field in ranges:
        status, message = check_range(field, patient.get(field))
        messages.append(message)
        blocked = blocked or status == "BLOCK"

    # Summarize only key teaching result.
    decision = "STOP" if blocked else "CONTINUE"
    warning_count = sum("unusual" in item for item in messages)
    print(f"Patient {patient['id']}: {decision}, warnings={warning_count}")

# Show one assumption behind the validation.
print("Assumption: adult height must be recorded in centimeters.")

```

### **1.3. Type Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_01_03.jpg?v=1781144909" width="250">



>* Ensure inputs match expected data types
>* Prevent unsafe or misleading clinical calculations

>* Mixed data sources cause type mismatches
>* Checks separate negatives, missing, and review cases

>* Explain type errors clearly for correction
>* Prevent unsafe calculations from mismatched inputs



```
#@title Python Code - Type Checks

# Type checks protect clinical calculation inputs.
# Each field should match expectations.
# Warnings guide safer human review.

# Define one small patient record.
patient_record = {
    "heart_rate": "eighty",
    "oxygen_saturation": 98,

    "symptom_present": "yes",
    "weight_kg": 70.5,
}

# Store expected Python types.
expected_types = {
    "heart_rate": int,
    "oxygen_saturation": int,

    "symptom_present": bool,
    "weight_kg": float,
}

# Prepare containers for findings.
valid_fields = []
warning_messages = []

# Check each required field safely.
for field_name, expected_type in expected_types.items():
    value = patient_record.get(field_name)

    # Missing values need review.
    if value is None:
        warning_messages.append(field_name + " is missing.")

    # Type mismatches should not be assumed.
    elif not isinstance(value, expected_type):
        actual_type = type(value).__name__

        expected_name = expected_type.__name__
        message = field_name + " expected " + expected_name
        message = message + ", received " + actual_type + "."
        warning_messages.append(message)

    # Matching types can proceed.
    else:
        valid_fields.append(field_name)

# Print concise validation summary.
print("Valid fields:", ", ".join(valid_fields))
print("Warnings found:", len(warning_messages))

# Show each warning clearly.
for warning_text in warning_messages:
    print("Warning:", warning_text)

# Explain the safe calculation rule.
ready = len(warning_messages) == 0
print("Ready for calculation:", ready)

```

## **2. Clinical Rule Scoring**

### **2.1. Point Based Rules**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_02_01.jpg?v=1781144904" width="250">



>* Assign points to matching clinical findings
>* Show each rule behind the total score

>* Use precise conditionals for awarding points
>* Store category scores in readable dictionaries

>* Handle missing inputs with warnings
>* Show assumptions behind each score



```
#@title Python Code - Point Based Rules

# Point rules make clinical scores transparent.
# Dictionaries keep scoring values organized.
# Validation prevents misleading quiet assumptions.

# Store patient data for a small example.
patient = {
    "age": 72,
    "temperature_c": 38.6,

    "respiratory_rate": 24,
    "systolic_bp": 88,
    "confusion": True,

    "oxygen_sat_percent": None,
    "symptom_severity": "moderate",
}

# Define category points using a dictionary.
severity_points = {
    "none": 0,
    "mild": 1,

    "moderate": 2,
    "severe": 3,
}

# Prepare result containers for transparency.
score = 0
reasons = []
warnings = []

# Check required numeric fields carefully.
required_fields = ["age", "temperature_c", "respiratory_rate", "systolic_bp"]
for field in required_fields:

    if patient.get(field) is None:
        warnings.append(f"Missing required value: {field}")

# Award points only when data are available.
if patient["age"] >= 65:
    score += 1
    reasons.append("Age at least 65: +1")

# Fever threshold is inclusive here.
if patient["temperature_c"] >= 38.0:
    score += 1
    reasons.append("Temperature at least 38.0 C: +1")

# Tachypnea rule uses a strict cutoff.
if patient["respiratory_rate"] > 22:
    score += 1
    reasons.append("Respiratory rate above 22: +1")

# Low blood pressure rule is inclusive.
if patient["systolic_bp"] <= 90:
    score += 2
    reasons.append("Systolic pressure 90 or lower: +2")

# Boolean findings can also add points.
if patient.get("confusion") is True:
    score += 2
    reasons.append("New confusion present: +2")

# Missing optional values should be reported.
if patient.get("oxygen_sat_percent") is None:
    warnings.append("Oxygen saturation missing; score is partial.")

# Dictionary lookup handles categorical scoring.
severity = patient.get("symptom_severity")
if severity in severity_points:
    score += severity_points[severity]

    reasons.append(f"Symptom severity {severity}: +{severity_points[severity]}")
else:
    warnings.append("Symptom severity category not recognized.")

# Print a concise clinical scoring report.
print(f"Point based score: {score}")
print("Contributing rules:")
print("; ".join(reasons))
print("Warnings:")
print("; ".join(warnings) if warnings else "None")
print("Educational example only; not a validated clinical tool.")

```

### **2.2. Threshold Logic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_02_02.jpg?v=1781144902" width="250">



>* Compare measurements against defined clinical cutoffs
>* Interpret thresholds within patient-specific context

>* Define threshold boundaries clearly
>* Avoid gaps, overlaps, and double counting

>* Check tiered thresholds in the right order
>* Explain results within the clinical context



```
#@title Python Code - Threshold Logic

# Threshold rules turn measurements into categories.
# Clear boundaries prevent gaps and overlap.
# Warnings communicate assumptions and clinical limitations.

# Store example patient measurements.
patient_values = {
    "temperature_c": 38.2,
    "oxygen_sat_percent": 91,
    "systolic_bp_mmhg": 88,
}

# Store threshold rules and warnings.
rules = {
    "temperature_c": {"high": 38.0, "warning": "Possible fever threshold met."},
    "oxygen_sat_percent": {"low": 92, "warning": "Low oxygen saturation threshold met."},
    "systolic_bp_mmhg": {"low": 90, "warning": "Low systolic pressure threshold met."},
}

# Create empty score and warning containers.
score = 0
warnings = []

# Validate each expected measurement exists.
for name, rule in rules.items():
    value = patient_values.get(name)

    # Skip missing values with a clear warning.
    if value is None:
        warnings.append(f"Missing value for {name}.")
        continue

    # Check higher-than-or-equal threshold rules.
    if "high" in rule and value >= rule["high"]:
        score += 1
        warnings.append(rule["warning"])

    # Check lower-than threshold rules.
    if "low" in rule and value < rule["low"]:
        score += 1
        warnings.append(rule["warning"])

# Convert score into a simple risk category.
if score == 0:
    category = "No threshold flags"
elif score == 1:
    category = "One threshold flag"
else:
    category = "Multiple threshold flags"

# Print concise clinical teaching results.
print("Threshold score:", score)
print("Category:", category)
print("Warnings:", "; ".join(warnings))
print("Note: screening thresholds support judgment, not diagnosis.")

```

### **2.3. Looped Score Logic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_02_03.jpg?v=1781144901" width="250">



>* Use loops for repeated clinical scoring checks
>* Reduce duplicated logic and scoring errors

>* Store scoring rules in structured dictionaries
>* Loop applies each rule consistently

>* Structured rules simplify score updates
>* Missing data requires cautious reporting



```
#@title Python Code - Looped Score Logic

# Looped scoring applies repeated clinical rules.
# Dictionaries keep criteria easy to inspect.
# Warnings separate missing data from normal findings.

# Define a small example patient record.
patient = {
    "age_years": 72,
    "heart_rate": 118,

    "respiratory_rate": 24,
    "oxygen_saturation": None,
    "confusion": True,
}

# Store rules as named dictionary entries.
rules = {
    "age_years": {"warning": 65, "points": 1},
    "heart_rate": {"warning": 110, "points": 2},

    "respiratory_rate": {"warning": 22, "points": 2},
    "oxygen_saturation": {"warning": 92, "points": 2},
    "confusion": {"warning": True, "points": 3},
}

# Prepare score and warning containers.
total_score = 0
triggered_rules = []
missing_rules = []

# Loop through every rule consistently.
for item, rule in rules.items():
    value = patient.get(item)

    # Missing values are unresolved, not normal.
    if value is None:
        missing_rules.append(item)
        continue

    # Oxygen saturation uses low threshold logic.
    if item == "oxygen_saturation":
        meets_rule = value <= rule["warning"]

    # Confusion uses a direct Boolean rule.
    elif item == "confusion":
        meets_rule = value is rule["warning"]

    # Other numeric rules use high threshold logic.
    else:
        meets_rule = value >= rule["warning"]

    # Add points only when the rule is met.
    if meets_rule:
        total_score += rule["points"]
        triggered_rules.append(item)

# Print compact teaching results.
print("Looped clinical score example")
print(f"Total score: {total_score}")
print(f"Triggered criteria: {triggered_rules}")
print(f"Missing criteria: {missing_rules}")
print("Warning: educational example, not validated clinical guidance.")

```

## **3. Clear Clinical Reporting**

### **3.1. Warnings and Assumptions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_03_01.jpg?v=1781144912" width="250">



>* Explain what results mean and miss
>* Show assumptions to avoid false confidence

>* Make warnings visible, specific, and meaningful
>* Highlight key uncertainties without overwhelming readers

>* State assumptions to support safe judgment
>* Use calculations as decision aids only



```
#@title Python Code - Warnings and Assumptions

# Clinical reports should explain calculation uncertainty.
# Warnings identify data and interpretation concerns.
# Assumptions clarify when results may mislead.

# Store one small patient example.
patient_age_years = 84
patient_weight_kg = 52
serum_creatinine_mg_dl = 1.1
creatinine_stable = False

# Store formula details and validation limits.
rule_limits = {"age_min": 18, "age_max": 80}
assumptions = []
warnings = []

# Add assumptions required by this estimate.
assumptions.append("Serum creatinine reflects stable kidney function.")
assumptions.append("Weight is current and recorded in kilograms.")
assumptions.append("Result supports review, not automatic dosing.")

# Check age against validated rule limits.
if patient_age_years > rule_limits["age_max"]:
    warnings.append("Age exceeds this rule's typical validation range.")

# Check weight for a plausible adult value.
if patient_weight_kg < 40 or patient_weight_kg > 200:
    warnings.append("Weight is outside a common adult plausibility range.")

# Check serum creatinine stability assumption.
if creatinine_stable is False:
    warnings.append("Creatinine is not stable, so estimate may mislead.")

# Calculate a simple teaching-only kidney estimate.
estimated_clearance = ((140 - patient_age_years) * patient_weight_kg) / (72 * serum_creatinine_mg_dl)

# Choose a simple dosing review category.
if estimated_clearance < 30:
    dose_category = "renal dose review likely needed"
else:
    dose_category = "standard review still required"

# Print a concise clinical report.
print("Clinical calculation report")
print(f"Estimated clearance: {estimated_clearance:.1f} mL/min")
print(f"Category: {dose_category}")
print("Warnings:")
print("; ".join(warnings) if warnings else "No warnings generated.")
print("Assumptions:")
print("; ".join(assumptions))

```

### **3.2. Assumptions and Limitations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_03_02.jpg?v=1781144913" width="250">



>* Show assumptions behind every clinical calculation
>* Hidden assumptions can lead to misuse

>* Calculations support, not replace, clinical judgment
>* Warn when patient context limits accuracy

>* Specific warnings prevent overconfident interpretation
>* State uncertainty to support safer communication



```
#@title Python Code - Assumptions and Limitations

# Clinical reports should explain calculation boundaries.
# Warnings make hidden assumptions visible.
# This example reports BMI responsibly.

# Define one example patient record.
patient = {
    "name": "Example adult",
    "age_years": 52,
    "height_cm": 170,

    "weight_kg": 92,
    "measured_today": False,
    "pregnant": False,
    "pediatric": False,
}

# Collect warnings in one list.
warnings = []
height_m = patient["height_cm"] / 100
bmi = patient["weight_kg"] / (height_m ** 2)

# Add assumption warnings when needed.
if patient["age_years"] < 18:
    warnings.append("BMI category assumes adult interpretation.")
if patient["measured_today"] is False:
    warnings.append("Height or weight may be outdated.")

# Add limitation warnings for context.
if patient["pregnant"] is True:
    warnings.append("BMI may not reflect pregnancy changes.")
if patient["weight_kg"] <= 0:
    warnings.append("Weight must be clinically plausible.")

# Choose a simple adult BMI category.
if bmi < 18.5:
    category = "underweight"
elif bmi < 25:
    category = "normal range"
elif bmi < 30:
    category = "overweight"
else:
    category = "obesity range"

# Print concise clinical reporting output.
print("Clear clinical report")
print(f"Patient: {patient['name']}")
print(f"BMI: {bmi:.1f} kg/m^2")
print(f"Category: {category}")

# State assumptions before listing warnings.
print("Assumption: adult BMI categories are being used.")
print("Limitation: BMI does not measure body composition.")
print("Warnings:")

# Print each warning, or a safe default.
for item in warnings:
    print(f"- {item}")
if not warnings:
    print("- No additional warnings were triggered.")

```

### **3.3. Explain Limitations Clearly**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_B/image_03_03.jpg?v=1781144915" width="250">



>* Show what each result can mean
>* Make input limitations and assumptions visible

>* State specific limits in meaningful language
>* Link each limit to clinical action

>* Use scores as clinical guidance, not certainty
>* State limits clearly near each result



```
#@title Python Code - Explain Limitations Clearly

# Clinical reports should explain calculation limits.
# Warnings help prevent overconfident interpretations.
# This example builds cautious score text.

# Store one compact simulated patient record.
patient = dict(age=82, weight_kg=68, creatinine=None)

# Record assumptions used by the calculation.
assumptions = dict(creatinine=1.2, source="estimated")

# Define a tiny rule based risk score.
score = 0
if patient["age"] >= 75:
    score = score + 1

if assumptions["creatinine"] >= 1.5:
    score = score + 1

# Convert score into a simple category.
category = "higher concern" if score >= 2 else "usual concern"

# Build warnings close to the result.
warnings = []
if patient["creatinine"] is None:
    warnings.append("Creatinine was missing; estimated value used.")

if assumptions["source"] == "estimated":
    warnings.append("Renal risk interpretation may be less reliable.")

if patient["age"] > 80:
    warnings.append("Older age may require broader clinical review.")

# Print a concise clinical style report.
print("Clinical calculation report")
print(f"Score: {score} ({category})")
print("Assumption: creatinine treated as 1.2 mg/dL")
print("Limitations and cautions:")

# Print each warning without excessive output.
for number, warning in enumerate(warnings, start=1):
    print(f"{number}. {warning}")

# End with a judgment reminder.
print("Use this result as decision support, not diagnosis.")

```
