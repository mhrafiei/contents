## **1. Reusable Medical Formulas**

### **1.1. BMI Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_01_01.jpg?v=1781143720" width="250">



>* BMI uses height and weight measurements
>* Functions make BMI calculations reusable

>* State expected BMI input units clearly
>* Consistent units prevent misleading results

>* Calculate BMI accurately without diagnosing alone
>* Use realistic tests and clear assumptions



```
#@title Python Code - BMI Functions

# BMI functions demonstrate reusable clinical calculations.
# Parameters clarify expected medical measurement units.
# Return values support testing and reuse.

def bmi_metric(weight_kg, height_m):
    """Return BMI from kilograms and metres."""
    if weight_kg <= 0 or height_m <= 0:
        raise ValueError("Weight and height must be positive.")

    bmi_value = weight_kg / (height_m ** 2)
    return round(bmi_value, 1)

# Define an imperial BMI function.
def bmi_imperial(weight_lb, height_in):
    """Return BMI from pounds and inches."""
    if weight_lb <= 0 or height_in <= 0:
        raise ValueError("Weight and height must be positive.")

    bmi_value = 703 * weight_lb / (height_in ** 2)
    return round(bmi_value, 1)

# Define simple clinical BMI labels.
def bmi_category(bmi_value):
    """Return a common adult BMI category."""
    if bmi_value < 18.5:
        return "underweight"

    if bmi_value < 25:
        return "healthy range"
    if bmi_value < 30:
        return "overweight range"

    return "obesity range"

# Test metric and imperial examples.
case_a = bmi_metric(weight_kg=70, height_m=1.75)
case_b = bmi_metric(weight_kg=82, height_m=1.80)
case_c = bmi_imperial(weight_lb=154, height_in=69)

# Display concise teaching outputs.
print("Metric adult A BMI:", case_a, bmi_category(case_a))
print("Metric adult B BMI:", case_b, bmi_category(case_b))
print("Imperial adult C BMI:", case_c, bmi_category(case_c))
print("Expected adult A value: about 22.9")
print("Functions reuse one formula safely.")

```

### **1.2. Medication Dose Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_01_02.jpg?v=1781143721" width="250">



>* Functions turn measurements into consistent medication doses
>* They support, not replace, clinical judgment

>* Identify required inputs and returned result
>* Keep units clear and outputs predictable

>* Dose functions improve consistency, not judgment
>* Review outputs within clinical context



```
#@title Python Code - Medication Dose Calculations

# Functions make medication calculations reusable.
# Clear parameters help prevent unit confusion.
# Example outputs still require clinical review.

# Define a daily weight based dose function.
def daily_weight_dose_mg(weight_kg, dose_mg_per_kg):
    """Return total daily dose in milligrams."""
    if weight_kg <= 0 or dose_mg_per_kg <= 0:
        raise ValueError("Weight and dose must be positive.")

    return weight_kg * dose_mg_per_kg

# Define a single dose divider function.
def single_dose_mg(total_daily_mg, doses_per_day):
    """Return milligrams for each scheduled dose."""
    if total_daily_mg <= 0 or doses_per_day <= 0:
        raise ValueError("Dose values must be positive.")

    return total_daily_mg / doses_per_day

# Define liquid volume from concentration.
def liquid_volume_ml(ordered_mg, concentration_mg_per_ml):
    """Return milliliters needed for an ordered dose."""
    if ordered_mg <= 0 or concentration_mg_per_ml <= 0:
        raise ValueError("Medication values must be positive.")

    return ordered_mg / concentration_mg_per_ml

# Store a realistic pediatric example.
patient_weight_kg = 18
recommended_mg_per_kg_day = 40
scheduled_doses_per_day = 2
concentration_mg_per_ml = 250 / 5

# Calculate daily and individual medication doses.
total_daily_dose = daily_weight_dose_mg(
    patient_weight_kg, recommended_mg_per_kg_day
)

each_dose = single_dose_mg(
    total_daily_dose, scheduled_doses_per_day
)

# Calculate measurable liquid volume per dose.
volume_per_dose = liquid_volume_ml(
    each_dose, concentration_mg_per_ml
)

# Print concise results for clinical interpretation.
print("Medication dose calculation example")
print(f"Weight: {patient_weight_kg} kg")
print(f"Daily dose: {total_daily_dose:.0f} mg/day")
print(f"Each dose: {each_dose:.0f} mg")
print(f"Volume per dose: {volume_per_dose:.1f} mL")
print("Verify limits, rounding, and patient factors.")

```

### **1.3. Dose Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_01_03.jpg?v=1781143724" width="250">



>* Patient measurements guide safe dose calculations
>* Functions make clinical rules reusable

>* Identify inputs and keep units consistent
>* Clarify dose type and clinical meaning

>* Functions standardize repeated dose calculations safely
>* Clear design supports clinical review



```
#@title Python Code - Dose Calculations

# Functions can clarify repeated medication calculations.
# Parameters describe measurements used by formulas.
# Returned values support safe checking steps.

def single_dose_mg(weight_kg, mg_per_kg, max_mg):
    """Calculate one weight based dose with maximum limit."""
    if weight_kg <= 0 or mg_per_kg <= 0:
        raise ValueError("Weight and dose rate must be positive.")

    calculated_mg = weight_kg * mg_per_kg
    limited_mg = min(calculated_mg, max_mg)
    return round(limited_mg, 1)

def daily_dose_mg(weight_kg, mg_per_kg_day):
    """Calculate total daily dose from body weight."""
    if weight_kg <= 0 or mg_per_kg_day <= 0:
        raise ValueError("Inputs must be positive clinical numbers.")

    total_daily_mg = weight_kg * mg_per_kg_day
    return round(total_daily_mg, 1)

def dose_each_time(total_daily_mg, doses_per_day):
    """Divide total daily dose across administrations."""
    if total_daily_mg <= 0 or doses_per_day <= 0:
        raise ValueError("Dose and frequency must be positive.")

    each_dose_mg = total_daily_mg / doses_per_day
    return round(each_dose_mg, 1)

# Store realistic example patient measurements.
child_weight_kg = 18.0
adult_weight_kg = 72.0

# Calculate a pediatric single antibiotic dose.
child_single_mg = single_dose_mg(child_weight_kg, 15, 500)
print(f"Child single dose: {child_single_mg} mg")

# Calculate an adult total daily dose.
adult_daily_mg = daily_dose_mg(adult_weight_kg, 20)
print(f"Adult total daily dose: {adult_daily_mg} mg/day")

# Convert daily dose into divided administrations.
adult_each_mg = dose_each_time(adult_daily_mg, 3)
print(f"Adult dose each time: {adult_each_mg} mg")

# Compare expected values for simple function tests.
print("Test child dose equals expected:", child_single_mg == 270.0)
print("Test adult each dose equals expected:", adult_each_mg == 480.0)

```

## **2. Reusable Function Design**

### **2.1. Clear Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_02_01.jpg?v=1781143714" width="250">



>* Name inputs clearly with required clinical context
>* Specify units and meanings to prevent confusion

>* Clear names help teams use functions safely
>* Specific measurements reduce clinical calculation errors

>* Use only clinically meaningful inputs
>* Keep parameter names and units consistent



```
#@title Python Code - Clear Parameters

# Clear parameters reduce clinical calculation mistakes.
# Names should show measurement units clearly.
# Return values make functions reusable.

def calculate_bmi(weight_kg, height_cm):
    """Return BMI using weight kilograms and height centimeters."""

    # Convert height before applying formula.
    height_m = height_cm / 100

    # Guard against impossible height values.
    if height_m <= 0:
        raise ValueError("height_cm must be positive")

    # Return a rounded reusable numeric result.
    return round(weight_kg / (height_m ** 2), 1)


def calculate_infusion_rate(volume_ml, time_hours):
    """Return infusion rate in milliliters per hour."""

    # Guard against impossible time values.
    if time_hours <= 0:
        raise ValueError("time_hours must be positive")

    # Return a rounded reusable numeric result.
    return round(volume_ml / time_hours, 1)


# Test realistic example patient measurements.
example_bmi = calculate_bmi(weight_kg=70, height_cm=175)

# Test realistic infusion order values.
example_rate = calculate_infusion_rate(volume_ml=500, time_hours=4)

# Display concise results for learners.
print("BMI result:", example_bmi, "kg/m^2")

# Display expected clinical infusion rate.
print("Infusion rate:", example_rate, "mL/hour")

# Show why parameter names matter.
print("Clear names show units before calculation.")

```

### **2.2. Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_02_02.jpg?v=1781143716" width="250">



>* Return calculated results for reuse
>* Keep calculation separate from later decisions

>* Match returns to calculation needs
>* Keep return formats predictable and consistent

>* Returned results enable realistic test checking
>* Clear returns support safer clinical workflows



```
#@title Python Code - Return Values

# Return values make calculations reusable.
# Clinical functions should avoid hidden printing.
# Tests compare returned results reliably.

# Define a reusable BMI function.
def calculate_bmi(weight_kg, height_m):
    """Return body mass index from kilograms and meters."""

    # Validate inputs before division.
    if weight_kg <= 0 or height_m <= 0:
        return None

    # Return the calculated clinical value.
    bmi_value = weight_kg / (height_m ** 2)
    return bmi_value

# Define an imperial helper function.
def inches_to_meters(height_inches):
    """Return meters converted from height in inches."""

    # Use standard inch conversion.
    meters_value = height_inches * 0.0254
    return meters_value

# Define a simple interpretation function.
def interpret_bmi(bmi_value):
    """Return a short BMI category label."""

    # Handle invalid calculation results.
    if bmi_value is None:
        return "invalid measurements"

    # Compare returned value with cutoffs.
    if bmi_value < 18.5:
        return "underweight"

    # Continue common adult BMI categories.
    if bmi_value < 25:
        return "healthy range"

    # Return final category for this example.
    return "above healthy range"

# Store returned values for reuse.
metric_bmi = calculate_bmi(70, 1.75)
imperial_height_m = inches_to_meters(68)
imperial_bmi = calculate_bmi(82, imperial_height_m)

# Test against expected clinical examples.
expected_metric_bmi = 22.86
expected_imperial_bmi = 27.48
metric_passed = round(metric_bmi, 2) == expected_metric_bmi
imperial_passed = round(imperial_bmi, 2) == expected_imperial_bmi

# Build concise output lines.
metric_line = f"Metric BMI: {metric_bmi:.2f}, {interpret_bmi(metric_bmi)}"
imperial_line = f"Imperial-height BMI: {imperial_bmi:.2f}, {interpret_bmi(imperial_bmi)}"
test_line = f"Tests passed: metric={metric_passed}, imperial={imperial_passed}"

# Print returned values after reuse.
print(metric_line)
print(imperial_line)
print(test_line)

```

### **2.3. Docstrings for Clarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_02_03.jpg?v=1781143718" width="250">



>* Explain function purpose, inputs, units, and outputs
>* Prevent misleading results and guide future users

>* Explain clinical context and calculation limits
>* Document units, assumptions, and output meaning

>* Docstrings make shared functions easier to verify
>* Readable documentation supports safer clinical software



```
#@title Python Code - Docstrings for Clarity

# Docstrings explain clinical functions clearly.
# Units must be documented carefully.
# Return values need plain descriptions.

# Define a documented BMI function.
def calculate_bmi(weight_kg, height_m):
    """Calculate adult body mass index.

    Parameters:
        weight_kg: Body weight in kilograms.
        height_m: Standing height in meters.

    Returns:
        BMI as a numeric index, not a diagnosis.
    """
    if weight_kg <= 0 or height_m <= 0:
        raise ValueError("Weight and height must be positive.")

    bmi_value = weight_kg / (height_m ** 2)
    return round(bmi_value, 1)

# Define a documented dose function.
def calculate_daily_dose(weight_kg, mg_per_kg_day):
    """Calculate a weight-based total daily medication dose.

    Parameters:
        weight_kg: Patient weight in kilograms.
        mg_per_kg_day: Ordered daily dose per kilogram.

    Returns:
        Total daily dose in milligrams.
    """
    if weight_kg <= 0 or mg_per_kg_day <= 0:
        raise ValueError("Dose inputs must be positive.")

    dose_value = weight_kg * mg_per_kg_day
    return round(dose_value, 1)

# Test functions with realistic examples.
patient_bmi = calculate_bmi(70, 1.75)
patient_dose = calculate_daily_dose(18, 15)

# Print concise teaching results.
print("BMI example:", patient_bmi, "kg/m^2")
print("Expected BMI near:", 22.9, "kg/m^2")
print("Daily dose example:", patient_dose, "mg/day")
print("Expected dose:", 270.0, "mg/day")
print("BMI docstring first line:", calculate_bmi.__doc__.strip().splitlines()[0])

```

```
#@title Python Code - Example Test Cases

# Clinical functions need realistic tests.
# Expected outputs document intended behavior.
# Boundary cases reveal calculation issues.

def bmi_category(weight_kg, height_m):
    """Return BMI and its adult category."""
    bmi_value = weight_kg / (height_m ** 2)

    if bmi_value < 18.5:
        category = "underweight"
    elif bmi_value < 25:
        category = "normal"

    elif bmi_value < 30:
        category = "overweight"
    else:
        category = "obesity"

    return round(bmi_value, 1), category


def pediatric_dose(weight_kg, mg_per_kg):
    """Return a weight based medication dose."""
    dose_mg = weight_kg * mg_per_kg
    return round(dose_mg, 1)

# Store realistic clinical example cases.
test_cases = [
    ("BMI ordinary adult", bmi_category(70, 1.75), (22.9, "normal")),
    ("BMI boundary adult", bmi_category(81, 1.80), (25.0, "overweight")),

    ("Pediatric dose", pediatric_dose(18, 10), 180.0),
]

# Compare each actual result with expectations.
summary = []
for case_name, actual, expected in test_cases:
    result = "PASS" if actual == expected else "CHECK"
    summary.append((case_name, actual, expected, result))

# Print compact teaching output.
print("Example clinical function tests:")
for case_name, actual, expected, result in summary:
    print(f"{case_name}: actual={actual}, expected={expected}, {result}")

# Count successful example tests.
passed_tests = sum(item[3] == "PASS" for item in summary)
print(f"Passed {passed_tests} of {len(summary)} example tests.")

```

### **3.2. Expected Outputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_03_02.jpg?v=1781143711" width="250">



>* Expected outputs show correct clinical results
>* Compare logic, units, and implementation

>* Use realistic clinically meaningful test cases
>* Check safe rounding and clear results

>* Verify outputs with independent trusted sources
>* Document assumptions, units, and rounding rules



```
#@title Python Code - Expected Outputs

# Expected outputs guide safer clinical function testing.
# Small examples make calculation checks easier.
# Independent values prevent trusting broken code.

# Define a reusable BMI calculation function.
def calculate_bmi(weight_kg, height_m):
    """Return body mass index from kilograms and meters."""
    if weight_kg <= 0 or height_m <= 0:
        raise ValueError("Measurements must be positive numbers")

    bmi_value = weight_kg / (height_m ** 2)
    return round(bmi_value, 1)

# Define a Celsius to Fahrenheit conversion function.
def celsius_to_fahrenheit(temp_c):
    """Return Fahrenheit temperature from Celsius temperature."""
    fahrenheit_value = (temp_c * 9 / 5) + 32
    return round(fahrenheit_value, 1)

# Store realistic clinical test cases.
test_cases = [
    ("Adult BMI", calculate_bmi, (70, 1.75), 22.9),
    ("Fever temp", celsius_to_fahrenheit, (39.0,), 102.2),
]

# Compare function outputs with expected outputs.
for case_name, function_used, inputs, expected in test_cases:
    actual = function_used(*inputs)
    passed = actual == expected
    print(f"{case_name}: expected {expected}, got {actual}, pass={passed}")

# Summarize why expected outputs matter.
print("Expected outputs check clinical logic and implementation together.")

```

### **3.3. Unexpected Clinical Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Medicine/Module_03/Lecture_A/image_03_03.jpg?v=1781143712" width="250">



>* Test missing, unusual, or impossible inputs
>* Separate rare clinical values from data errors

>* Check math and clinical meaning
>* Reject unsafe or misleading outputs

>* Test realistic edge cases and special populations
>* Check data quality before trusting outputs



```
#@title Python Code - Unexpected Clinical Inputs

# Clinical functions need unexpected input testing.
# This example checks BMI inputs safely.
# Each case shows expected clinical behavior.

def calculate_bmi(weight_kg, height_m):
    """Return BMI after checking clinical input boundaries."""

    # Reject missing measurements before calculating anything.
    if weight_kg is None or height_m is None:
        return "Invalid: missing weight or height"

    # Reject impossible or likely unit-mismatched values.
    if weight_kg <= 0 or height_m <= 0:
        return "Invalid: nonpositive weight or height"

    # Flag heights that are implausible in meters.
    if height_m < 0.4 or height_m > 2.7:
        return "Invalid: height may use wrong units"

    # Flag weights outside broad human plausibility limits.
    if weight_kg < 0.5 or weight_kg > 700:
        return "Invalid: weight outside broad limits"

    # Return a rounded BMI for accepted inputs.
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)

# Store realistic and unexpected clinical test cases.
test_cases = [
    ("Typical adult", 70, 1.75, 22.9),
    ("Zero height", 70, 0, "Invalid"),
    ("Centimeters entered", 70, 175, "Invalid"),
    ("Small neonate", 3.2, 0.50, 12.8),
    ("Missing weight", None, 1.65, "Invalid"),
]

# Print a compact test summary for learners.
print("Unexpected input tests for BMI function:")

# Compare each actual result with expected behavior.
for label, weight, height, expected in test_cases:
    result = calculate_bmi(weight, height)

    # Treat any invalid message as the expected invalid outcome.
    if expected == "Invalid":
        passed = isinstance(result, str) and result.startswith("Invalid")
    else:
        passed = result == expected

    # Print one concise line for each test case.
    status = "PASS" if passed else "CHECK"
    print(f"{label}: {result} -> {status}")

# End with one reusable safe calculation example.
print("Safe function avoids misleading clinical outputs.")

```
