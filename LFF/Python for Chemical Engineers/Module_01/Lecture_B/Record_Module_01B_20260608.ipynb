## **1. Engineering Variables**

### **1.1. Flow Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1780976432" width="250">



>* Flow rate measures material moving over time
>* Variables and arithmetic structure engineering calculations

>* Separate volume flow from mass flow
>* Use density and arithmetic to connect flows

>* Use flow arithmetic for engineering checks
>* Name quantities clearly to avoid mistakes



```
#@title Python Code - Flow Rate

# Variables can represent measured engineering quantities.
# Arithmetic expressions connect physical process ideas.
# Flow calculations require clear units.

# Define volumetric flow rates in liters per minute.
inlet_a_lpm = 12.5
inlet_b_lpm = 7.8
outlet_lpm = 18.9

# Combine inlet streams using ordinary addition.
total_inlet_lpm = inlet_a_lpm + inlet_b_lpm
net_accumulation_lpm = total_inlet_lpm - outlet_lpm

# Convert liquid volume flow to mass flow.
density_water_kg_per_liter = 0.997
mass_flow_kg_per_min = total_inlet_lpm * density_water_kg_per_liter

# Show key calculated flow quantities.
print("Total inlet flow:", total_inlet_lpm, "L/min")
print("Outlet flow:", outlet_lpm, "L/min")
print("Net accumulation:", round(net_accumulation_lpm, 2), "L/min")

# Use comparison to interpret tank behavior.
tank_level_message = "rising" if net_accumulation_lpm > 0 else "falling or steady"
print("Tank level is likely:", tank_level_message)

# Demonstrate an Imperial pipe velocity calculation.
pipe_area_ft2 = 0.0491
flow_ft3_per_min = 1.6
velocity_ft_per_min = flow_ft3_per_min / pipe_area_ft2

# Print a rounded velocity result.
print("Pipe velocity:", round(velocity_ft_per_min, 1), "ft/min")

```

### **1.2. Temperature Variables**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1780976433" width="250">



>* Temperature affects many process properties
>* Named variables give numbers engineering context

>* Track temperature units and scales carefully.
>* Convert values for accurate engineering calculations.

>* Use temperature arithmetic to describe process behavior
>* Clear variables support reliable advanced modeling



```
#@title Python Code - Temperature Variables

# Temperature variables clarify engineering calculations.
# Names connect numbers to process equipment.
# Unit conversions require careful arithmetic.

# Store a sensor temperature in Celsius.
feed_temperature_celsius = 85.0

# Convert Celsius to Kelvin for calculations.
feed_temperature_kelvin = feed_temperature_celsius + 273.15

# Store a reactor set point in Celsius.
reactor_setpoint_celsius = 95.0

# Calculate temperature rise across equipment.
temperature_rise_celsius = reactor_setpoint_celsius - feed_temperature_celsius

# Store a safety limit in Celsius.
safe_limit_celsius = 110.0

# Calculate margin below the safe limit.
safety_margin_celsius = safe_limit_celsius - reactor_setpoint_celsius

# Convert Celsius to Fahrenheit for operators.
feed_temperature_fahrenheit = (feed_temperature_celsius * 9 / 5) + 32

# Display the engineering temperature variables.
print("Feed temperature, deg C:", feed_temperature_celsius)
print("Feed temperature, K:", feed_temperature_kelvin)
print("Reactor set point, deg C:", reactor_setpoint_celsius)
print("Temperature rise, deg C:", temperature_rise_celsius)
print("Safety margin, deg C:", safety_margin_celsius)
print("Feed temperature, deg F:", feed_temperature_fahrenheit)

```

### **1.3. Process Pressure**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1780976430" width="250">



>* Pressure describes fluid force in equipment
>* Clear names and units prevent ambiguity

>* Calculate pressure changes, averages, and conversions
>* Relate pressure to other process variables

>* Distinguish absolute pressure from gauge pressure
>* Use clear names and consistent calculations



```
#@title Python Code - Process Pressure

# Variables can represent measured process pressures.
# Arithmetic expressions convert and compare pressures.
# Clear names help distinguish pressure types.

# Store a gauge pressure reading.
gauge_pressure_kpa = 250.0
# Store local atmospheric pressure.
atmospheric_pressure_kpa = 101.325

# Convert gauge pressure to absolute pressure.
absolute_pressure_kpa = gauge_pressure_kpa + atmospheric_pressure_kpa
# Store an outlet pressure after pipe losses.
outlet_pressure_kpa = 310.0

# Calculate a pressure drop across equipment.
pressure_drop_kpa = absolute_pressure_kpa - outlet_pressure_kpa
# Convert absolute pressure into bar units.
absolute_pressure_bar = absolute_pressure_kpa / 100.0

# Convert absolute pressure into psi units.
absolute_pressure_psi = absolute_pressure_kpa * 0.145038
# Summarize pressure values with readable labels.
print(f"Gauge pressure: {gauge_pressure_kpa:.1f} kPa")

print(f"Atmospheric pressure: {atmospheric_pressure_kpa:.3f} kPa")
print(f"Absolute pressure: {absolute_pressure_kpa:.3f} kPa")
print(f"Pressure drop: {pressure_drop_kpa:.3f} kPa")

print(f"Absolute pressure: {absolute_pressure_bar:.3f} bar")
print(f"Absolute pressure: {absolute_pressure_psi:.2f} psi")

```

## **2. Python Data Types**

### **2.1. Integer Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780976435" width="250">



>* Integers represent whole-number counts or steps.
>* They mark discrete, not continuous, quantities.

>* Integers support predictable whole-number calculations
>* Use decimals for continuously varying quantities

>* Integers organize labels, trials, and data positions
>* Meaning determines measured value or identifier



```
#@title Python Code - Integer Values

# This example introduces integer data values.
# Integer values represent whole engineering counts.
# Python can inspect each variable type.

# Store whole-number equipment and experiment counts.
tray_count = 18
reactor_count = 3
sample_count = 12
iteration_count = 50

# Store continuous values with decimal parts.
flow_rate_lpm = 24.5
temperature_c = 100.0
pipe_length_ft = 30.75

# Show values and their Python types.
print("tray_count:", tray_count, type(tray_count).__name__)
print("reactor_count:", reactor_count, type(reactor_count).__name__)
print("flow_rate_lpm:", flow_rate_lpm, type(flow_rate_lpm).__name__)

# Integer arithmetic keeps whole counts when appropriate.
total_samples = sample_count * reactor_count
print("total_samples:", total_samples, type(total_samples).__name__)

# Division can create a floating-point result.
samples_per_reactor = sample_count / reactor_count
print("samples_per_reactor:", samples_per_reactor, type(samples_per_reactor).__name__)

# Integer division keeps only complete groups.
complete_batches = sample_count // reactor_count
remaining_samples = sample_count % reactor_count
print("complete_batches:", complete_batches, "remaining:", remaining_samples)

# Built-in int converts suitable decimal values.
rounded_temperature_label = int(temperature_c)
print("temperature label:", rounded_temperature_label, type(rounded_temperature_label).__name__)

# Identifiers can be integers without representing measurements.
reactor_id = 100
print("reactor_id:", reactor_id, type(reactor_id).__name__)

```

### **2.2. String Data Types**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780976437" width="250">



>* Strings store text, labels, and descriptions
>* Separate text identifiers from numerical quantities

>* Strings add context to engineering data
>* Python separates text from calculable numbers

>* Strings format and communicate text-based results
>* Convert text numbers before doing calculations



```
#@title Python Code - String Data Types

# Strings store descriptive engineering information.
# They are different from numeric values.
# Text operations support readable engineering reports.

# Define text labels used near calculations.
chemical_name = "ethanol"
stream_label = "feed stream"
unit_label = "kg/h"

# Define a numeric value separately.
mass_flow_rate = 125.5
numeric_type = type(mass_flow_rate).__name__
text_type = type(unit_label).__name__

# Combine strings and numbers for output.
report_line = f"{stream_label.title()}: {mass_flow_rate} {unit_label} of {chemical_name}."
print(report_line)
print("Flow value type:", numeric_type)
print("Unit label type:", text_type)

# Show that text digits are still strings.
equipment_label = "R101"
equipment_number = 101
print("Equipment label type:", type(equipment_label).__name__)
print("Equipment number type:", type(equipment_number).__name__)

# Convert numeric text before arithmetic.
concentration_text = "2.50"
concentration_value = float(concentration_text)
adjusted_value = concentration_value * 1.10
print("Adjusted concentration:", adjusted_value, "mol/L")

```

### **2.3. Boolean Types**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780976439" width="250">



>* Booleans store true-or-false conditions
>* They support yes-or-no engineering decisions

>* Comparisons produce true-or-false results
>* Booleans connect measurements to decisions

>* Booleans control program decisions and workflows
>* Booleans represent truth, not quantities or text



```
#@title Python Code - Boolean Types

# Boolean values support engineering decisions.
# Comparisons create true or false results.
# Logical checks combine process safety conditions.

# Define measured operating quantities.
temperature_c = 88.0
pressure_bar = 4.8
valve_open = True

# Define allowable process limits.
max_temperature_c = 90.0
max_pressure_bar = 5.0
minimum_flow_lpm = 12.0

# Compare measurements against process limits.
temperature_safe = temperature_c <= max_temperature_c
pressure_safe = pressure_bar <= max_pressure_bar
flow_lpm = 10.5

# Combine Boolean checks using and.
flow_sufficient = flow_lpm >= minimum_flow_lpm
unit_safe = temperature_safe and pressure_safe and flow_sufficient
needs_attention = not unit_safe

# Print Boolean values and their types.
print("temperature_safe:", temperature_safe, type(temperature_safe))
print("pressure_safe:", pressure_safe, type(pressure_safe))
print("flow_sufficient:", flow_sufficient, type(flow_sufficient))

# Show stored Boolean states directly.
print("valve_open:", valve_open, type(valve_open))
print("unit_safe:", unit_safe, type(unit_safe))
print("needs_attention:", needs_attention, type(needs_attention))

# Use a Boolean in a decision.
status_message = "Check process" if needs_attention else "Continue operation"
print("recommended action:", status_message)

```

## **3. Built In Functions**

### **3.1. Rounding Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780976426" width="250">



>* Round results to meaningful precision
>* Make numbers easier to communicate

>* Choose decimal places based on context
>* Round final reported results, not calculations

>* Rounding changes display, not actual accuracy
>* Match reported precision to data quality



```
#@title Python Code - Rounding Results

# Built-in round clarifies engineering numerical results.
# Chemical calculations often produce extra decimals.
# Round displayed results after calculations finish.

# Store unrounded reactor and process quantities.
reactor_volume_liters = 12.34567
feed_flow_liters_per_minute = 2.71828
residence_time_minutes = reactor_volume_liters / feed_flow_liters_per_minute

# Round the computed residence time for reporting.
rounded_time_one = round(residence_time_minutes, 1)
rounded_time_two = round(residence_time_minutes, 2)

# Show full and rounded numerical results.
print("Full residence time:", residence_time_minutes)
print("Rounded to 1 decimal:", rounded_time_one)
print("Rounded to 2 decimals:", rounded_time_two)

# Round a conversion fraction for percent reporting.
conversion_fraction = 0.87654321
conversion_percent = conversion_fraction * 100
rounded_conversion = round(conversion_percent, 2)

# Display a clear process-performance result.
print("Conversion percent:", rounded_conversion)

# Demonstrate rounding to the nearest whole unit.
sample_count_estimate = 14.6
rounded_samples = round(sample_count_estimate)

# Display a whole-number planning quantity.
print("Estimated samples needed:", rounded_samples)

```

### **3.2. Math Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780976428" width="250">



>* Math functions simplify common engineering calculations
>* They make physical quantities clearer

>* Use functions for absolute, minimum, maximum values
>* Sum data to support engineering judgment

>* Match functions to data type and meaning
>* Interpret results within engineering workflows



```
#@title Python Code - Math Functions

# Built-in math functions support engineering calculations.
# This example uses small process measurements.
# Results stay readable for beginner learners.

# Store measured and target concentrations.
measured_concentration = 0.094

target_concentration = 0.100

# Use abs for unsigned concentration error.
concentration_error = abs(measured_concentration - target_concentration)

# Store reactor temperatures in Celsius.
temperature_readings_c = [72.5, 75.0, 73.8, 76.2]

# Use min and max for limits.
coldest_temperature = min(temperature_readings_c)

hottest_temperature = max(temperature_readings_c)

# Store inlet mass flow rates.
inlet_flows_kg_per_hr = [120.0, 85.5, 44.5]

# Use sum for total feed.
total_feed_kg_per_hr = sum(inlet_flows_kg_per_hr)

# Use pow for squared velocity.
velocity_m_per_s = 3.2

velocity_squared = pow(velocity_m_per_s, 2)

# Print concise engineering results.
print("Concentration error:", concentration_error, "mol/L")

print("Coldest temperature:", coldest_temperature, "C")

print("Hottest temperature:", hottest_temperature, "C")

print("Total inlet flow:", total_feed_kg_per_hr, "kg/hr")

print("Velocity squared:", velocity_squared, "m^2/s^2")

```

### **3.3. Text Operations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780976425" width="250">



>* Use text functions to organize engineering labels
>* Create clear outputs with numbers and units

>* Standardize messy text before use
>* Add clear labels to engineering results

>* Connect Python with external data sources
>* Format text for clear engineering records



```
#@title Python Code - Text Operations

# Built-in functions help inspect engineering text.
# Text labels make numerical results clearer.
# This example formats a reactor report.

# Store text with inconsistent spacing and capitalization.
raw_material = "  ethanol  "

# Clean text using common string operations.
material_clean = raw_material.strip().title()

# Store numerical values for a small calculation.
temperature_c = 78.37
flow_rate_l_min = 2.5

# Convert numbers into readable text labels.
temperature_text = str(round(temperature_c, 1))
flow_text = str(round(flow_rate_l_min, 2))

# Measure text length using a built-in function.
material_length = len(material_clean)

# Combine text pieces into clear engineering messages.
report_line = material_clean + " at " + temperature_text + " C"

# Build another message containing flow information.
flow_line = "Feed flow is " + flow_text + " L/min"

# Print compact results for the learner.
print("Clean material:", material_clean)
print("Material characters:", material_length)
print("Report line:", report_line)
print("Flow message:", flow_line)

```
