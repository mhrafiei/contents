## **1. Conditional Safety Logic**

### **1.1. If Statements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780977556" width="250">



>* If statements guide process-based decisions
>* Actions run only when conditions are met

>* Match conditions to physical process meaning
>* Choose thresholds with sound engineering judgment

>* Write safety if statements clearly and cautiously
>* Use engineering judgment, not just code



```
#@title Python Code - If Statements

# This script demonstrates basic reactor safety logic.
# An if statement checks one engineering condition.
# The result changes only when needed.

# Define the measured reactor temperature.
reactor_temperature_c = 92.0

# Define the maximum acceptable temperature limit.
safe_temperature_limit_c = 85.0

# Compare the measurement with the limit.
if reactor_temperature_c > safe_temperature_limit_c:
    action = "Increase cooling and alert the operator."

# Report the decision to the learner.
print("Reactor temperature:", reactor_temperature_c, "deg C")
print("Safety limit:", safe_temperature_limit_c, "deg C")

# Show the action selected by the condition.
print("Decision:", action)

```

### **1.2. Conditional Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1780977554" width="250">



>* Compare process data with safe operating limits
>* Trigger actions when conditions become unsafe

>* Match checks to engineering rules
>* Consider units, data quality, and consequences

>* Use clear, conservative safety checks
>* Capture trends, thresholds, and engineering judgment



```
#@title Python Code - Conditional Checks

# Conditional checks encode safe process decisions.
# This example compares temperature and pressure.
# Each branch prints one clear action.

# Define current reactor operating readings.
temperature_c = 92.0
pressure_kpa = 240.0
cooling_flow_lpm = 18.0

# Define safety limits from operating procedure.
max_temperature_c = 90.0
max_pressure_kpa = 250.0
min_cooling_flow_lpm = 20.0

# Start with a normal safety message.
action = "Continue normal operation"
reason = "all monitored values acceptable"

# Check the most urgent shutdown condition.
if temperature_c > max_temperature_c and cooling_flow_lpm < min_cooling_flow_lpm:
    action = "Initiate controlled shutdown"
    reason = "high temperature with low cooling flow"

# Check a pressure alarm condition.
elif pressure_kpa >= max_pressure_kpa:
    action = "Trigger high-pressure alarm"
    reason = "pressure reached allowable limit"

# Check a temperature warning condition.
elif temperature_c > max_temperature_c:
    action = "Reduce heat input"
    reason = "temperature exceeded safe limit"

# Report compact results for operators.
print("Conditional safety check result")
print(f"Temperature: {temperature_c} C")
print(f"Pressure: {pressure_kpa} kPa")
print(f"Cooling flow: {cooling_flow_lpm} L/min")
print(f"Action: {action}")
print(f"Reason: {reason}")

```

### **1.3. Safety Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1780977557" width="250">



>* Define safe operating limits in code
>* Use thresholds to flag unsafe process data

>* Combine checks to capture real safety context
>* Support decisions without replacing safety systems

>* Make safety checks clear and conservative
>* Treat missing or uncertain data as unsafe



```
#@title Python Code - Safety Conditions

# Encode safety boundaries using clear conditions.
# Treat missing measurements as unsafe information.
# Combine process facts before declaring safety.

# Define a small reactor monitoring snapshot.
temperature_c = 132
pressure_kpa = None
cooling_flow_lpm = 48
agitator_running = True

# Store named safety limits for readability.
max_temperature_c = 130
max_pressure_kpa = 450
min_cooling_flow_lpm = 40
safe_status_text = "SAFE"

# Start with a conservative default decision.
decision = "REVIEW"
reason = "Uncertain condition requires engineering attention."

# Check missing pressure before comparing limits.
if pressure_kpa is None:
    decision = "ALERT"
    reason = "Pressure reading is missing, not automatically safe."

# Check temperature against the operating limit.
elif temperature_c >= max_temperature_c:
    decision = "ALERT"
    reason = "Temperature is at or above the allowed limit."

# Check pressure against vessel rating logic.
elif pressure_kpa >= max_pressure_kpa:
    decision = "ALERT"
    reason = "Pressure is at or above the allowed limit."

# Check combined cooling and agitation conditions.
elif cooling_flow_lpm < min_cooling_flow_lpm or not agitator_running:
    decision = "ALERT"
    reason = "Cooling or mixing is not adequate."

# Otherwise classify the snapshot as acceptable.
else:
    decision = safe_status_text
    reason = "All required measurements satisfy the rules."

# Print a compact operating summary.
print("Reactor safety decision:", decision)
print("Reason:", reason)
print("Temperature C:", temperature_c)
print("Pressure kPa:", pressure_kpa)
print("Cooling flow L/min:", cooling_flow_lpm)
print("Agitator running:", agitator_running)

```

## **2. Looping Over Conditions**

### **2.1. For Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780977544" width="250">



>* Repeat calculations across many operating conditions
>* Shorter scripts with fewer repeated mistakes

>* Loop through each operating condition
>* Compare results and find process trends

>* Record and compare results consistently
>* Design loops to preserve relevant outcomes



```
#@title Python Code - For Loops

# For loops repeat engineering calculations.
# Each pass uses one condition.
# Results can support process decisions.

# Define feed temperatures in degrees Celsius.
temperatures_C = [300, 320, 340, 360, 380]

# Define a safe conversion limit.
minimum_conversion = 0.55

# Prepare a list for loop results.
conversion_results = []

# Loop over each operating temperature.
for temperature_C in temperatures_C:

    # Estimate conversion with a simple trend.
    conversion = 0.40 + 0.0012 * temperature_C

    # Store each temperature and conversion.
    conversion_results.append((temperature_C, conversion))

# Print a compact table heading.
print("Temperature C | Conversion | Status")

# Loop over stored calculation results.
for temperature_C, conversion in conversion_results:

    # Check the engineering decision rule.
    if conversion >= minimum_conversion:

        # Label acceptable operating conditions.
        status = "acceptable"

    else:

        # Label conditions below specification.
        status = "below spec"

    # Print one concise result row.
    print(f"{temperature_C:13.0f} | {conversion:10.2f} | {status}")

# Find the best condition after looping.
best_temperature, best_conversion = max(conversion_results, key=lambda item: item[1])

# Print the best operating point.
print(f"Best case: {best_temperature} C gives conversion {best_conversion:.2f}")

```

### **2.2. While Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780977547" width="250">



>* Repeat calculations until a condition is met
>* Useful when iteration counts are unknown

>* Update changing values each loop cycle
>* Match loop logic to the process

>* Add limits to prevent endless looping
>* Stop when close enough or not converging



```
#@title Python Code - While Loops

# While loops repeat until conditions change.
# Engineers update states during each iteration.
# Safeguards prevent runaway engineering calculations.

# Define tank filling parameters in metric units.
target_volume_L = 25.0
current_volume_L = 2.0
flow_rate_L_min = 3.5
time_min = 0.0

# Define a maximum time safety limit.
max_time_min = 10.0
time_step_min = 1.0
iteration_count = 0
status_message = "target not reached"

# Repeat while target remains unmet.
while current_volume_L < target_volume_L:
    current_volume_L += flow_rate_L_min * time_step_min
    time_min += time_step_min
    iteration_count += 1

    # Stop if safety time is exceeded.
    if time_min >= max_time_min:
        status_message = "stopped by safety limit"
        break

# Record success if target condition became false.
if current_volume_L >= target_volume_L:
    status_message = "target reached safely"

# Summarize the loop result compactly.
print("While-loop tank filling example")
print(f"Iterations completed: {iteration_count}")
print(f"Final time: {time_min:.1f} min")
print(f"Final volume: {current_volume_L:.1f} L")
print(f"Loop status: {status_message}")

```

### **2.3. Range Iteration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780977546" width="250">



>* Repeat calculations across planned operating ranges
>* Evaluate each condition in a controlled sweep

>* Choose range limits and step sizes carefully
>* Balance detail, efficiency, and engineering judgment

>* Treat loop results as sampled process behavior
>* Choose ranges using physical limits and sensitivity



```
#@title Python Code - Range Iteration

# Range loops evaluate planned operating conditions.
# Each pass represents one reactor temperature.
# Small tables reveal trends without repetition.

# Define the temperature sweep in Celsius.
temperatures_c = range(300, 381, 20)

# Store compact results for later review.
conversion_results = []

# Loop through each planned temperature.
for temperature_c in temperatures_c:
    rate_constant = 0.0015 * (temperature_c - 280)
    conversion = 1 - pow(2.71828, -rate_constant * 8)

    # Keep conversion as a percentage.
    conversion_percent = 100 * conversion
    conversion_results.append((temperature_c, conversion_percent))

# Print a short engineering table.
print("Temperature sweep for a small plug flow reactor")
print("Temp C | Conversion percent")

# Display each sampled operating condition.
for temperature_c, conversion_percent in conversion_results:
    print(f"{temperature_c:6d} | {conversion_percent:18.1f}")

# Identify the first temperature meeting target conversion.
target_percent = 60
met_target = False

# Check the results in order.
for temperature_c, conversion_percent in conversion_results:
    if conversion_percent >= target_percent:
        print(f"Target first met near {temperature_c} C")
        met_target = True
        break

# Report if the tested range missed the target.
if met_target is False:
    print("Target not met in this temperature range")

```

## **3. Debugging Basics**

### **3.1. Identify Logic Errors**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780977551" width="250">



>* Code runs, but decisions may be wrong
>* Compare outputs with intended engineering rules

>* Match conditions to real engineering rules
>* Check boundaries, comparisons, and overlaps

>* Test known typical, extreme, and boundary cases
>* Trace each value through decision branches



```
#@title Python Code - Identify Logic Errors

# Logic errors can hide inside working scripts.
# Boundary tests reveal many engineering decision mistakes.
# Plain language rules guide corrected conditions.

# Store temperature tests in degrees Celsius.
test_temperatures_c = [79.0, 80.0, 80.1, 95.0]

# State the intended reactor safety rule.
print("Rule: safe when T <= 80 C; otherwise caution.")

# Define a buggy classifier with reversed boundary logic.
def buggy_label(temperature_c):
    if temperature_c < 80.0:
        return "safe"
    else:
        return "caution"

# Define the corrected classifier using the rule.
def corrected_label(temperature_c):
    if temperature_c <= 80.0:
        return "safe"
    else:
        return "caution"

# Compare both decisions against known boundary cases.
print("T(C) | buggy | corrected")
for temperature_c in test_temperatures_c:
    bad = buggy_label(temperature_c)
    good = corrected_label(temperature_c)

    # Print compact results for each test value.
    print(f"{temperature_c:4.1f} | {bad:7s} | {good}")

# Highlight the boundary case that exposes the error.
print("Logic error found at exactly 80.0 C.")

```

### **3.2. Loop Logic Errors**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780977549" width="250">



>* Loops may repeat too many or few times
>* Check intended conditions and boundary cases

>* Check loop updates and stopping conditions
>* Place calculations and saved totals correctly

>* Trace first, middle, and final loop passes
>* Check boundaries and loop update behavior



```
#@title Python Code - Loop Logic Errors

# Debug loop boundaries carefully.
# Chemical cases need exact coverage.
# Trace repeated calculations explicitly.

# No extra installations are required.

# Define intended reactor temperatures in Celsius.
temperatures_c = [300, 320, 340, 360]

# Calculate simple conversion values for each case.
correct_results = []

# Process every intended temperature exactly once.
for temp_c in temperatures_c:
    conversion = 0.45 + 0.001 * (temp_c - 300)

    # Store each temperature and conversion pair.
    correct_results.append((temp_c, conversion))

# Demonstrate an off-by-one loop mistake.
buggy_results = []

# This range accidentally skips the last index.
for index in range(0, len(temperatures_c) - 1):
    temp_c = temperatures_c[index]

    # Store the incomplete loop calculation.
    conversion = 0.45 + 0.001 * (temp_c - 300)

    # Save buggy results for comparison.
    buggy_results.append((temp_c, conversion))

# Print a compact debugging summary.
print("Intended temperatures:", temperatures_c)

# Show which temperatures the buggy loop processed.
buggy_temps = [item[0] for item in buggy_results]

# Show which temperatures the correct loop processed.
correct_temps = [item[0] for item in correct_results]

# Report the loop coverage comparison.
print("Buggy loop processed:", buggy_temps)
print("Correct loop processed:", correct_temps)

# Identify missing operating conditions defensively.
missing_temps = sorted(set(correct_temps) - set(buggy_temps))

# Print the detected loop logic error.
print("Missing temperatures:", missing_temps)

# Confirm the corrected loop covers all cases.
print("Correct result count:", len(correct_results))

```

### **3.3. Print Debugging**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780977552" width="250">



>* Print values to trace program paths
>* Check whether correct logic branches run

>* Print only at key decision points
>* Trace program flow against intended logic

>* Use clear diagnostics, then clean them up
>* Test assumptions with quick printed evidence



```
#@title Python Code - Print Debugging

# Print debugging reveals control flow decisions.
# Chemical rules often use conditional branches.
# Diagnostic messages should stay focused.

# Define reactor temperatures in degrees Celsius.
temperatures_c = [72, 81, 89, 96]

# Define the intended safety limit.
safety_limit_c = 90

# Count how many operating points are safe.
safe_count = 0

# Announce the debugging run clearly.
print("Debugging reactor temperature decisions")

# Loop through a small set of conditions.
for index, temperature_c in enumerate(temperatures_c, start=1):

    # Show values before the decision.
    print(f"Run {index}: T={temperature_c} C, limit={safety_limit_c} C")

    # Check which branch the script enters.
    if temperature_c <= safety_limit_c:

        # Confirm the safe branch was reached.
        print("  Branch: safe operation")

        # Update the safe operating counter.
        safe_count = safe_count + 1

    # Handle temperatures above the limit.
    else:

        # Confirm the warning branch was reached.
        print("  Branch: safety warning")

# Summarize the observed control flow.
print(f"Safe operating points counted: {safe_count}")

```
