## **1. Sensible Heat Calculations**

### **1.1. Heat Capacity Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1780983929" width="250">



>* Heat capacity measures energy for temperature change
>* Sensible heat links temperature change to duty

>* Choose mass or molar heat capacity basis
>* Match heat capacity units to flow units

>* Heat capacity varies with conditions
>* Use clear, updateable values in Python



```
#@title Python Code - Heat Capacity Basics

# Sensible heat links flow, heat capacity, and temperature.
# Unit consistency prevents common energy balance mistakes.
# This example compares mass and molar bases.

# Define a compact sensible heat calculation function.
def sensible_heat(flow_rate, heat_capacity, delta_temperature):
    return flow_rate * heat_capacity * delta_temperature

# Store stream data with clear basis labels.
water_flow_kg_s = 2.0
water_cp_kj_kg_c = 4.18
water_delta_c = 35.0

# Calculate water heating duty on a mass basis.
water_duty_kw = sensible_heat(
    water_flow_kg_s,
    water_cp_kj_kg_c,
    water_delta_c,
)

# Store nitrogen data on a molar basis.
nitrogen_flow_kmol_h = 12.0
nitrogen_cp_kj_kmol_c = 29.1
nitrogen_delta_c = -80.0

# Calculate nitrogen cooling duty in kilojoules per hour.
nitrogen_duty_kj_h = sensible_heat(
    nitrogen_flow_kmol_h,
    nitrogen_cp_kj_kmol_c,
    nitrogen_delta_c,
)

# Convert nitrogen duty for easier comparison.
nitrogen_duty_kw = nitrogen_duty_kj_h / 3600.0

# Check that chosen heat capacities are positive.
if water_cp_kj_kg_c <= 0 or nitrogen_cp_kj_kmol_c <= 0:
    raise ValueError("Heat capacities must be positive values")

# Print concise teaching results.
print("Sensible heat uses: duty = flow * Cp * delta T")
print(f"Water heater duty: {water_duty_kw:.1f} kW")
print(f"Nitrogen cooler duty: {nitrogen_duty_kw:.1f} kW")
print("Positive duty means heating; negative duty means cooling.")
print("Always match Cp basis with the stream flow basis.")

```

### **1.2. Temperature Difference**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1780983927" width="250">



>* Temperature change drives sensible heat calculations
>* Identify inlet and outlet temperatures carefully

>* Temperature change sign shows heating or cooling
>* Use final minus initial consistently

>* Distinguish temperature intervals from absolute temperatures
>* Display temperatures clearly to catch errors



```
#@title Python Code - Temperature Difference

# Sensible heat starts with temperature difference.
# Direction shows heating or cooling behavior.
# Small checks prevent process calculation mistakes.

# Store inlet and outlet temperatures.
initial_temperature_c = 25.0
final_temperature_c = 80.0

# Calculate final minus initial temperature.
delta_temperature_c = final_temperature_c - initial_temperature_c

# Temperature intervals match in Celsius and kelvin.
delta_temperature_k = delta_temperature_c

# Interpret the sign for process direction.
if delta_temperature_c > 0:
    process_direction = "heating"
else:
    process_direction = "cooling"

# Check for zero temperature difference separately.
if delta_temperature_c == 0:
    process_direction = "no sensible heat change"

# Print a compact readable energy-balance check.
print("Sensible Heat Temperature Difference")
print(f"Initial temperature: {initial_temperature_c:.1f} °C")
print(f"Final temperature: {final_temperature_c:.1f} °C")
print(f"Delta T: {delta_temperature_c:.1f} °C")
print(f"Delta T: {delta_temperature_k:.1f} K")
print(f"Process interpretation: {process_direction}")

```

### **1.3. Unit Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1780983931" width="250">



>* Match units before sensible heat calculations
>* Convert time basis for final heat duty

>* Separate temperatures from temperature changes
>* Scale Fahrenheit intervals before heat calculations

>* Convert inputs to one consistent unit basis
>* Explicit units improve checking and communication



```
#@title Python Code - Unit Conversion

# Convert units before sensible heat calculations.
# Temperature differences need careful scale handling.
# This example reports heater duty in kilowatts.

# Store plant data in original units.
flow_lb_per_min = 120.0
cp_btu_per_lb_F = 0.80
t_in_F = 70.0
t_out_F = 180.0

# Check that outlet temperature is hotter.
if t_out_F <= t_in_F:
    raise ValueError("Outlet temperature must be higher.")

# Convert mass flow to kilograms per second.
lb_to_kg = 0.45359237
seconds_per_minute = 60.0
flow_kg_per_s = flow_lb_per_min * lb_to_kg / seconds_per_minute

# Convert heat capacity to kilojoules per kilogram kelvin.
btu_to_kJ = 1.055056
F_interval_per_K = 1.8
cp_kJ_per_kg_K = cp_btu_per_lb_F * btu_to_kJ / lb_to_kg * F_interval_per_K

# Convert the Fahrenheit temperature difference correctly.
delta_T_F = t_out_F - t_in_F
delta_T_K = delta_T_F / F_interval_per_K

# Calculate sensible heat duty in kilowatts.
heat_duty_kW = flow_kg_per_s * cp_kJ_per_kg_K * delta_T_K
wrong_heat_duty_kW = flow_kg_per_s * cp_kJ_per_kg_K * delta_T_F

# Report compact results for engineering checking.
print("Sensible heat unit conversion example")
print(f"Mass flow: {flow_kg_per_s:.3f} kg/s")
print(f"Heat capacity: {cp_kJ_per_kg_K:.3f} kJ/(kg K)")
print(f"Temperature difference: {delta_T_K:.2f} K")
print(f"Correct heater duty: {heat_duty_kW:.1f} kW")
print(f"If Fahrenheit interval is misused: {wrong_heat_duty_kW:.1f} kW")
print(f"Error factor: {wrong_heat_duty_kW / heat_duty_kW:.2f}")

```

## **2. Heat Duty**

### **2.1. Heater Heat Duty**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1780983938" width="250">



>* Heater duty is required energy input
>* Depends on stream amount and states

>* Use heat capacity for sensible heating
>* Report heater duty as positive input

>* Include latent heat during phase changes
>* Break heating paths into clear steps



```
#@title Python Code - Heater Heat Duty

# Estimate heater duty using sensible heat.
# This example keeps units explicit.
# Positive duty means heat added.

# Define stream data using metric engineering units.
mass_flow_kg_s = 2.5
inlet_temp_c = 25.0
outlet_temp_c = 95.0
cp_kj_kg_c = 4.18

# Check temperatures before calculating heat duty.
if outlet_temp_c <= inlet_temp_c:
    raise ValueError("Outlet temperature must exceed inlet temperature.")

# Calculate temperature rise and heater duty.
delta_temp_c = outlet_temp_c - inlet_temp_c
heat_duty_kw = mass_flow_kg_s * cp_kj_kg_c * delta_temp_c

# Convert duty into common plant reporting units.
heat_duty_btu_hr = heat_duty_kw * 3412.142
heat_duty_mmbtu_hr = heat_duty_btu_hr / 1_000_000

# Estimate utility power including heater efficiency.
heater_efficiency = 0.85
required_utility_kw = heat_duty_kw / heater_efficiency

# Print a compact readable calculation summary.
print("Heater heat duty example")
print(f"Mass flow: {mass_flow_kg_s:.2f} kg/s")
print(f"Temperature rise: {delta_temp_c:.1f} deg C")
print(f"Stream heat duty: {heat_duty_kw:.1f} kW")
print(f"Stream heat duty: {heat_duty_mmbtu_hr:.3f} MMBtu/hr")
print(f"Utility power at 85% efficiency: {required_utility_kw:.1f} kW")

```

### **2.2. Cooler Duty**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780983935" width="250">



>* Heat removed determines required cooling service
>* Use consistent signs for clear sizing

>* Identify stream phases and temperature changes
>* Include sensible and latent heat effects

>* Match cooling utility to process limits
>* Check safety, phases, and physical consistency



```
#@title Python Code - Cooler Duty

# Estimate cooler duty for one liquid stream.
# Use sensible heat with consistent metric units.
# Report both process and utility sign conventions.

# Define known stream conditions.
mass_flow_kg_s = 2.5
heat_capacity_kj_kg_C = 4.18
inlet_temperature_C = 85.0
outlet_temperature_C = 35.0

# Check that cooling is physically requested.
if outlet_temperature_C >= inlet_temperature_C:
    raise ValueError("Outlet temperature must be lower for cooling.")

# Calculate the stream temperature change.
delta_temperature_C = outlet_temperature_C - inlet_temperature_C
process_duty_kw = mass_flow_kg_s * heat_capacity_kj_kg_C * delta_temperature_C
heat_removed_kw = -process_duty_kw

# Estimate cooling water requirement.
water_cp_kj_kg_C = 4.18
water_temperature_rise_C = 10.0
cooling_water_kg_s = heat_removed_kw / (water_cp_kj_kg_C * water_temperature_rise_C)

# Print compact engineering results.
print("Cooler duty example: hot liquid product.")
print(f"Process heat duty = {process_duty_kw:.1f} kW.")
print(f"Heat removed by utility = {heat_removed_kw:.1f} kW.")
print(f"Cooling water needed = {cooling_water_kg_s:.2f} kg/s.")
print("Negative process duty means heat leaves the process stream.")

```

### **2.3. Reactor Heat Duty**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1780983937" width="250">



>* Reactions may release or absorb heat
>* Duty combines temperature and reaction heat effects

>* Separate temperature effects from reaction heat
>* Base reaction heat on actual conversion

>* State heat duty sign convention clearly
>* Check assumptions for safe reactor design



```
#@title Python Code - Reactor Heat Duty

# Estimate heat duty for a reactor.
# Combine sensible and reaction heat terms.
# Positive duty means heat is added.

# Store stream and reaction data.
flow_mol_s = 2.0
conversion_fraction = 0.60
cp_j_mol_k = 75.0

# Define inlet and outlet temperatures.
t_in_celsius = 25.0
t_out_celsius = 80.0
temperature_change_k = t_out_celsius - t_in_celsius

# Check values before calculating duty.
valid_conversion = 0.0 <= conversion_fraction <= 1.0
positive_flow = flow_mol_s > 0.0
positive_cp = cp_j_mol_k > 0.0

# Stop early if assumptions are invalid.
if not (valid_conversion and positive_flow and positive_cp):
    raise ValueError("Check flow, conversion, and heat capacity.")

# Calculate physical heating of material.
sensible_watt = flow_mol_s * cp_j_mol_k * temperature_change_k
reaction_enthalpy_j_mol = -50000.0
reaction_extent_mol_s = flow_mol_s * conversion_fraction

# Calculate chemical heat from reaction extent.
reaction_watt = reaction_extent_mol_s * reaction_enthalpy_j_mol
total_duty_watt = sensible_watt + reaction_watt
total_duty_kw = total_duty_watt / 1000.0

# Interpret the sign using stated convention.
if total_duty_kw >= 0.0:
    duty_message = "heat added to reactor"
else:
    duty_message = "heat removed from reactor"

# Print a compact engineering summary.
print("Reactor heat duty estimate")
print(f"Sensible term: {sensible_watt / 1000.0:.2f} kW")
print(f"Reaction term: {reaction_watt / 1000.0:.2f} kW")
print(f"Total duty: {total_duty_kw:.2f} kW")
print(f"Interpretation: {abs(total_duty_kw):.2f} kW {duty_message}")

```

## **3. Script Structure**

### **3.1. Input Validation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780983921" width="250">



>* Validate inputs before energy balance calculations
>* Catch missing, wrong-unit, or unrealistic values

>* Separate validation improves readability and maintenance
>* Checks reveal assumptions and valid conditions

>* Clear messages identify invalid inputs quickly
>* Early validation builds trustworthy calculations



```
#@title Python Code - Input Validation

# Validate inputs before energy calculations.
# Clear checks protect engineering assumptions.
# Friendly messages help script users.

# Define one heater case dictionary.
case = {
    "label": "water heater",
    "mass_flow_kg_s": 2.5,
    "cp_kJ_kgK": 4.18,
    "tin_C": 25.0,
    "tout_C": 80.0,
}

# Define validation for heater inputs.
def validate_heater_case(data):
    required = ["label", "mass_flow_kg_s", "cp_kJ_kgK", "tin_C", "tout_C"]

    # Check for any missing keys.
    missing = [key for key in required if key not in data]
    if missing:
        raise ValueError(f"Missing required inputs: {missing}")

    # Read values after key validation.
    mass_flow = data["mass_flow_kg_s"]
    heat_capacity = data["cp_kJ_kgK"]
    inlet_temp = data["tin_C"]
    outlet_temp = data["tout_C"]

    # Check physically meaningful numeric ranges.
    if mass_flow <= 0:
        raise ValueError("Mass flow must be positive kg/s.")

    if not 0.5 <= heat_capacity <= 6.0:
        raise ValueError("Heat capacity looks unreasonable for liquids.")

    # Check temperature ordering for heater.
    if outlet_temp <= inlet_temp:
        raise ValueError("Heater outlet must exceed inlet temperature.")

    return True

# Validate before performing calculations.
validation_passed = validate_heater_case(case)

# Calculate sensible heat duty safely.
delta_temp = case["tout_C"] - case["tin_C"]
heat_duty_kW = case["mass_flow_kg_s"] * case["cp_kJ_kgK"] * delta_temp

# Display compact engineering results.
print(f"Validation passed: {validation_passed}")
print(f"Stream label: {case['label']}")
print(f"Temperature change: {delta_temp:.1f} K")
print(f"Heat duty: {heat_duty_kW:.1f} kW")

```

### **3.2. Calculation Blocks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780983923" width="250">



>* Organize calculations in a clear sequence
>* Separate blocks for easier review and updates

>* Separate each block by calculation purpose
>* Inspect blocks to find heat-duty errors

>* Use blocks to inspect intermediate results
>* Improve reliability, auditing, and script extension



```
#@title Python Code - Calculation Blocks

# Organize energy balance work into readable blocks.
# Check intermediate values before final duties.
# Use clear names for engineering quantities.

# Block one defines process stream inputs.
stream_name = "liquid feed"
mass_flow_kg_s = 1.80
inlet_temp_c = 25.0

# Continue inputs with target and property data.
outlet_temp_c = 85.0
heat_capacity_kj_kg_c = 4.18
max_reasonable_kw = 600.0

# Block two checks input values.
valid_flow = mass_flow_kg_s > 0.0
valid_capacity = heat_capacity_kj_kg_c > 0.0
valid_temperatures = outlet_temp_c != inlet_temp_c

# Combine checks before calculations proceed.
inputs_are_valid = all([valid_flow, valid_capacity, valid_temperatures])
if not inputs_are_valid:
    raise ValueError("Check flow, heat capacity, and temperatures.")

# Block three calculates sensible heating terms.
temperature_change_c = outlet_temp_c - inlet_temp_c
heat_duty_kw = mass_flow_kg_s * heat_capacity_kj_kg_c * temperature_change_c
operation_label = "heater" if heat_duty_kw > 0.0 else "cooler"

# Block four checks engineering reasonableness.
absolute_duty_kw = abs(heat_duty_kw)
duty_is_large = absolute_duty_kw > max_reasonable_kw
warning_text = "review inputs" if duty_is_large else "reasonable size"

# Block five reports compact intermediate results.
print(f"Stream: {stream_name}")
print(f"Temperature change: {temperature_change_c:.1f} C")
print(f"Estimated {operation_label} duty: {heat_duty_kw:.1f} kW")
print(f"Reasonableness check: {warning_text}")

```

### **3.3. Error Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780983925" width="250">



>* Check calculations for physical meaning
>* Verify inputs, assumptions, and sign conventions

>* Check results against physical behavior
>* Use estimates to catch unreasonable duties

>* Place checks near related calculations
>* Use clear warnings for troubleshooting



```
#@title Python Code - Error Checks

# Energy balance scripts need careful checks.
# Clear warnings make calculations more trustworthy.
# This example checks heater inputs.

# Store stream data in readable variables.
stream_name = "water feed stream"
mass_flow_kg_s = 1.8
heat_capacity_kj_kgC = 4.18

# Store temperatures using expected Celsius units.
inlet_temperature_C = 25.0
outlet_temperature_C = 85.0
intended_operation = "heater"

# Create an empty warning list.
warnings_found = []
temperature_change_C = outlet_temperature_C - inlet_temperature_C
heat_duty_kW = mass_flow_kg_s * heat_capacity_kj_kgC * temperature_change_C

# Check that flow and heat capacity are positive.
if mass_flow_kg_s <= 0:
    warnings_found.append("Mass flow must be positive.")
if heat_capacity_kj_kgC <= 0:
    warnings_found.append("Heat capacity must be positive.")

# Check temperature direction for intended operation.
if intended_operation == "heater" and temperature_change_C <= 0:
    warnings_found.append("Heater should increase stream temperature.")
if intended_operation == "cooler" and temperature_change_C >= 0:
    warnings_found.append("Cooler should decrease stream temperature.")

# Check for suspicious temperature magnitudes.
if inlet_temperature_C > 200 or outlet_temperature_C > 200:
    warnings_found.append("Temperature seems high for liquid water.")
if abs(temperature_change_C) > 150:
    warnings_found.append("Temperature change may be unusually large.")

# Check duty sign convention consistency.
if intended_operation == "heater" and heat_duty_kW <= 0:
    warnings_found.append("Heater duty should be positive.")
if intended_operation == "cooler" and heat_duty_kW >= 0:
    warnings_found.append("Cooler duty should be negative.")

# Print compact calculation results.
print("Energy balance error check example")
print(f"Stream: {stream_name}")
print(f"Delta T: {temperature_change_C:.1f} C")
print(f"Heat duty: {heat_duty_kW:.1f} kW")

# Report whether any checks raised concerns.
if warnings_found:
    print("Warnings found:")
    print("; ".join(warnings_found))
else:
    print("All checks passed for this simple case.")

```
