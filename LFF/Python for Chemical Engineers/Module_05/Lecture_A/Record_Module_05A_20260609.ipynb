## **1. Importing Process Data**

### **1.1. CSV File Reading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_01_01.jpg?v=1780984924" width="250">



>* CSV files store process measurements simply
>* Python converts text rows into usable data

>* Recognize CSV rows, commas, and headers
>* Convert text fields into usable numbers

>* Check CSV files for messy real data
>* Clean imports support reliable engineering conclusions



```
#@title Python Code - CSV File Reading

# Read process CSV data safely.
# Convert text measurements into numbers.
# Summarize values for engineering interpretation.

import csv
from pathlib import Path

# Store tiny process data locally.
csv_text = """time_min,inlet_C,outlet_C,flow_L_min,pressure_kPa
0,72.1,55.4,10.2,18.5
1,73.0,56.2,10.1,18.7
2,72.6,55.9,,18.6
3,74.2,57.1,10.4,19.0
"""

# Write a small CSV file.
file_path = Path("heat_exchanger_readings.csv")
file_path.write_text(csv_text, encoding="utf-8")

# Prepare containers for clean rows.
clean_rows = []
missing_count = 0

# Open the CSV file for reading.
with file_path.open("r", newline="", encoding="utf-8") as file_object:
    reader = csv.DictReader(file_object)

    # Convert each text field carefully.
    for row in reader:
        if row["flow_L_min"].strip() == "":
            missing_count += 1
            continue

        # Build one cleaned measurement record.
        clean_rows.append({
            "time_min": int(row["time_min"]),
            "inlet_C": float(row["inlet_C"]),
            "outlet_C": float(row["outlet_C"]),
            "flow_L_min": float(row["flow_L_min"]),
            "pressure_kPa": float(row["pressure_kPa"]),
        })

# Validate that usable data were imported.
if len(clean_rows) == 0:
    raise ValueError("No complete process rows were imported.")

# Compute simple descriptive summaries.
inlet_values = [row["inlet_C"] for row in clean_rows]
outlet_values = [row["outlet_C"] for row in clean_rows]
pressure_values = [row["pressure_kPa"] for row in clean_rows]

# Calculate average cooling and pressure.
avg_inlet = sum(inlet_values) / len(inlet_values)
avg_outlet = sum(outlet_values) / len(outlet_values)
avg_pressure = sum(pressure_values) / len(pressure_values)

# Interpret the imported process data.
cooling_drop = avg_inlet - avg_outlet
print(f"Imported complete rows: {len(clean_rows)}")
print(f"Rows skipped for missing flow: {missing_count}")
print(f"Average inlet temperature: {avg_inlet:.1f} C")
print(f"Average outlet temperature: {avg_outlet:.1f} C")
print(f"Average cooling drop: {cooling_drop:.1f} C")
print(f"Average pressure drop: {avg_pressure:.1f} kPa")

```

### **1.2. Reading Text Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_01_02.jpg?v=1780984928" width="250">



>* Text files store many process measurements
>* Python reads file contents line by line

>* Check line structure and separators
>* Skip headers, blanks, and notes

>* Convert text lines into numeric process data
>* Check units, variables, and unusual entries



```
#@title Python Code - Reading Text Files

# This script reads a small process text file.
# It demonstrates line inspection and value conversion.
# Built-in tools support transparent data importing.

# Create a tiny reactor temperature text file.
file_text = """Reactor R-101 temperature log
Units: degC
Minute,Temperature
0,25.0
1,26.4
2,27.1
3,28.0
4,28.5
"""

# Write the example data into a file.
with open("reactor_temperatures.txt", "w") as file_handle:
    file_handle.write(file_text)

# Prepare containers for imported measurement values.
minutes = []
temperatures = []

# Read lines and skip descriptive headers.
with open("reactor_temperatures.txt", "r") as file_handle:
    all_lines = file_handle.readlines()

# Convert comma separated text into numbers.
for line in all_lines[3:]:
    clean_line = line.strip()
    if clean_line:
        parts = clean_line.split(",")
        minutes.append(int(parts[0]))
        temperatures.append(float(parts[1]))

# Compute simple statistics from imported data.
average_temp = sum(temperatures) / len(temperatures)
minimum_temp = min(temperatures)
maximum_temp = max(temperatures)

# Print a compact engineering summary.
print("Imported records:", len(temperatures))
print("First minute and temperature:", minutes[0], temperatures[0])
print("Average temperature, degC:", round(average_temp, 2))
print("Temperature range, degC:", minimum_temp, "to", maximum_temp)
print("Conclusion: reactor warmed steadily during this short run.")

```

### **1.3. Naming Data Columns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_01_03.jpg?v=1780984926" width="250">



>* Clear column names make datasets understandable
>* Specific labels prevent measurement confusion

>* Use descriptive names with process context
>* Keep naming consistent to prevent errors

>* Include units and measurement context clearly
>* Correct labels make statistics meaningful



```
#@title Python Code - Naming Data Columns

# Clear names make imported data easier.
# Chemical measurements need context and units.
# Built-in tools can rename text columns.

import csv
from io import StringIO

# Simulate a small process text file.
raw_text = """time,temp,press,conc\n0,348.2,2.05,0.104\n5,349.1,2.11,0.098\n10,347.8,2.08,0.101\n"""

# Read rows using built-in csv tools.
reader = csv.reader(StringIO(raw_text))
old_names = next(reader)
data_rows = list(reader)

# Map vague names to engineering names.
new_names = [
    "time_min",
    "reactor_temperature_K",
    "reactor_pressure_bar_abs",
    "product_concentration_mol_L",
]

# Validate header count before renaming.
if len(old_names) != len(new_names):
    raise ValueError("Header count does not match names")

# Build dictionaries using improved names.
records = [dict(zip(new_names, row)) for row in data_rows]

# Convert selected measurement columns to floats.
temperatures = [float(row["reactor_temperature_K"]) for row in records]
pressures = [float(row["reactor_pressure_bar_abs"]) for row in records]

# Compute two simple descriptive summaries.
avg_temperature = sum(temperatures) / len(temperatures)
max_pressure = max(pressures)

# Show concise evidence of clearer meaning.
print("Original names:", old_names)
print("Improved names:", new_names)
print("First renamed record:", records[0])
print("Average reactor temperature K:", round(avg_temperature, 2))
print("Maximum absolute pressure bar:", round(max_pressure, 2))

```

## **2. Descriptive Statistics**

### **2.1. Mean Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_02_01.jpg?v=1780984919" width="250">



>* Mean summarizes repeated process measurements.
>* Compare averages with targets or past performance.

>* Mean works best for stable, even data
>* Outliers can distort engineering interpretation

>* Means connect data to engineering decisions
>* Check context, variation, and unusual measurements



```
#@title Python Code - Mean Measurements

# This example summarizes repeated reactor measurements.
# The mean estimates typical process behavior.
# Extreme readings can shift engineering conclusions.

# No additional package installation is required.

# Store repeated reactor temperature measurements.
temperatures_c = [80.1, 79.8, 80.3, 80.0, 79.9, 80.2]

# Check that the dataset contains observations.
valid_count = len(temperatures_c) > 0

# Compute the mean using built-in Python.
mean_temp_c = sum(temperatures_c) / len(temperatures_c)

# Define the engineering target temperature.
target_temp_c = 80.0

# Compare the mean measurement with target.
difference_c = mean_temp_c - target_temp_c

# Add one unusual sensor reading.
temperatures_with_spike = temperatures_c + [86.5]

# Compute the shifted mean after the spike.
spike_mean_c = sum(temperatures_with_spike) / len(temperatures_with_spike)

# Print a concise process data summary.
print("Original readings:", len(temperatures_c))

# Print the normal operating average.
print("Mean temperature without spike:", round(mean_temp_c, 2), "°C")

# Print the deviation from target.
print("Difference from target:", round(difference_c, 2), "°C")

# Print the average including the spike.
print("Mean temperature with spike:", round(spike_mean_c, 2), "°C")

# Interpret the engineering meaning briefly.
print("Interpretation: the spike pulls the mean upward.")

# Confirm safe nonempty dataset handling.
print("Data valid for mean calculation:", valid_count)

```

### **2.2. Range and Spread**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_02_02.jpg?v=1780984917" width="250">



>* Spread shows process stability beyond averages
>* Wide variation can signal operational problems

>* Range shows the full measurement span
>* Standard deviation describes typical variation

>* Spread reveals risks beyond process averages
>* Context helps judge meaningful variation



```
#@title Python Code - Range and Spread

# Range shows total measurement span.
# Spread describes typical process variation.
# Engineers compare both before deciding.

# Import statistics for built-in calculations.
import statistics as stats

# Store reactor outlet temperatures, degrees Celsius.
temperatures_c = [
    350.2, 349.8, 350.5, 351.0,
    349.6, 350.1, 350.4, 352.8,
]

# Validate enough measurements for spread calculations.
if len(temperatures_c) < 2:
    raise ValueError("Need at least two measurements")

# Calculate basic descriptive statistics.
minimum_c = min(temperatures_c)
maximum_c = max(temperatures_c)
data_range_c = maximum_c - minimum_c

# Calculate center and typical spread.
mean_c = stats.mean(temperatures_c)
stdev_c = stats.stdev(temperatures_c)

# Count samples outside one standard deviation.
lower_limit_c = mean_c - stdev_c
upper_limit_c = mean_c + stdev_c
outside_count = sum(
    value < lower_limit_c or value > upper_limit_c
    for value in temperatures_c
)

# Print compact engineering summary lines.
print(f"Measurements analyzed: {len(temperatures_c)}")
print(f"Minimum and maximum: {minimum_c:.1f} to {maximum_c:.1f} °C")
print(f"Range: {data_range_c:.1f} °C")
print(f"Mean: {mean_c:.1f} °C")
print(f"Sample standard deviation: {stdev_c:.2f} °C")
print(f"Outside mean ± one stdev: {outside_count} measurements")

# Interpret spread using a simple engineering threshold.
if data_range_c <= 2.0:
    conclusion = "narrow spread, likely stable operation"
else:
    conclusion = "wide spread, investigate disturbances"

# Print the engineering interpretation.
print(f"Conclusion: {conclusion}.")

```

### **2.3. Data Distribution**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_02_03.jpg?v=1780984921" width="250">



>* Distribution shows patterns beyond the average
>* Extreme values may reveal process problems

>* Patterns show balance, skew, or clustering
>* Outliers require engineering judgment

>* Distribution shows when averages are reliable
>* Patterns reveal operating issues and regimes



```
#@title Python Code - Data Distribution

# Explore distributions in process measurements.
# Use small synthetic reactor data.
# Compare steady and upset behavior.

# No extra installations are required.
import random
import statistics
import matplotlib.pyplot as plt

# Make random values repeatable.
random.seed(42)
steady_temperatures = [random.gauss(85.0, 1.2) for _ in range(40)]
upset_temperatures = [random.gauss(85.0, 1.2) for _ in range(35)]

# Add occasional high-temperature process upsets.
upset_temperatures.extend([90.8, 91.4, 92.1, 93.0, 94.2])
all_lengths = [len(steady_temperatures), len(upset_temperatures)]
assert min(all_lengths) >= 10

# Summarize center and spread.
def summarize(label, values):
    mean_value = statistics.mean(values)
    median_value = statistics.median(values)
    stdev_value = statistics.stdev(values)
    range_value = max(values) - min(values)

    # Print compact engineering summary.
    print(f"{label}: mean={mean_value:.2f} C, median={median_value:.2f} C")
    print(f"{label}: stdev={stdev_value:.2f} C, range={range_value:.2f} C")
    return mean_value, median_value, stdev_value, range_value

# Compare both measurement distributions.
steady_summary = summarize("Steady reactor", steady_temperatures)
upset_summary = summarize("Reactor with upsets", upset_temperatures)

# Interpret distribution shape using summaries.
print("Similar means can hide very different temperature distributions.")
print("Larger spread and high tail suggest possible process instability.")

# Plot one distribution comparison.
plt.figure(figsize=(7, 4))
plt.hist(steady_temperatures, bins=8, alpha=0.7, label="Steady")
plt.hist(upset_temperatures, bins=8, alpha=0.7, label="With upsets")

# Label the engineering plot clearly.
plt.xlabel("Outlet temperature, degrees C")
plt.ylabel("Number of readings")
plt.title("Distribution of reactor outlet temperatures")
plt.legend()

# Show the compact distribution plot.
plt.tight_layout()
plt.show()

```

## **3. Engineering Data Interpretation**

### **3.1. Contextual Conclusions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_03_01.jpg?v=1780984932" width="250">



>* Link statistics to process context
>* Judge impacts on safety, quality, efficiency

>* Link statistics to practical engineering risks
>* Check variability, extremes, and process limits

>* Check data limits before drawing conclusions
>* Link statistics to process-aware actions



```
#@title Python Code - Contextual Conclusions

# Summaries need practical engineering context.
# Statistics alone cannot prove safety.
# Thresholds help connect numbers to action.

import statistics as stats

# Store small reactor temperature measurements.
temperatures_c = [78.2, 79.1, 80.4, 81.0, 82.6, 83.1, 84.0]

# Define the normal target and caution limit.
target_c = 80.0
caution_limit_c = 83.0

# Compute simple descriptive statistics.
avg_temp_c = stats.mean(temperatures_c)
stdev_temp_c = stats.stdev(temperatures_c)
max_temp_c = max(temperatures_c)

# Count observations above the caution limit.
high_count = sum(value > caution_limit_c for value in temperatures_c)
total_count = len(temperatures_c)

# Build a contextual engineering conclusion.
if avg_temp_c <= caution_limit_c and high_count == 0:
    conclusion = "stable operation near expected conditions"
else:
    conclusion = "average acceptable, but excursions need investigation"

# Print a compact engineering summary.
print(f"Average temperature: {avg_temp_c:.1f} °C")
print(f"Standard deviation: {stdev_temp_c:.1f} °C")
print(f"Maximum temperature: {max_temp_c:.1f} °C")
print(f"Readings above {caution_limit_c:.1f} °C: {high_count} of {total_count}")
print(f"Contextual conclusion: {conclusion}")

```

### **3.2. Trends**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_03_02.jpg?v=1780984934" width="250">



>* Trends reveal process changes over time
>* Patterns help identify possible equipment issues

>* Link trends to real process causes
>* Use judgment before drawing conclusions

>* Check trends against noise and supporting data
>* State uncertainty, causes, and next actions



```
#@title Python Code - Trends

# Trends turn measurements into engineering process stories.
# Small datasets can reveal gradual process drift.
# Summary statistics support cautious engineering conclusions.

# Import built-in tools for text data.
import csv, io, statistics

# Store a tiny process dataset as text.
data_text = """day,outlet_temp_C,steam_kg_h
1,82.4,510
2,82.1,511
3,81.9,509
4,81.5,510
5,81.2,512
6,80.9,511
7,80.6,510
"""

# Read the text dataset safely.
reader = csv.DictReader(io.StringIO(data_text))
rows = list(reader)

# Validate the dataset before calculations.
if len(rows) < 2:
    raise ValueError("Need at least two measurements.")

# Convert text values into numeric lists.
days = [int(row["day"]) for row in rows]
temps = [float(row["outlet_temp_C"]) for row in rows]
steam = [float(row["steam_kg_h"]) for row in rows]

# Compute simple descriptive statistics.
mean_temp = statistics.mean(temps)
range_temp = max(temps) - min(temps)
change_temp = temps[-1] - temps[0]
mean_steam = statistics.mean(steam)

# Estimate a simple trend slope.
day_span = days[-1] - days[0]
slope = change_temp / day_span

# Interpret the trend using process context.
if slope < -0.20:
    message = "Temperature shows a meaningful downward trend."
else:
    message = "Temperature appears approximately stable."

# Print concise engineering summaries.
print(f"Mean outlet temperature: {mean_temp:.1f} C")
print(f"Temperature range: {range_temp:.1f} C")
print(f"Temperature trend: {slope:.2f} C per day")
print(f"Mean steam flow: {mean_steam:.0f} kg/h")
print(message)
print("Stable steam flow suggests checking fouling or instrumentation.")

```

### **3.3. Quality Conclusions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_05/Lecture_A/image_03_03.jpg?v=1780984930" width="250">



>* Use summaries to judge process reliability
>* Consider both averages and variability

>* Compare statistics with process expectations
>* Use full data patterns, not favorites

>* Recognize uncertainty and data limitations
>* Base decisions on balanced evidence



```
#@title Python Code - Quality Conclusions

# Interpret summaries for engineering quality decisions.
# Built-in tools handle small process datasets.
# Statistics become evidence for action.

import csv
import io
import statistics as stats

# Store a tiny purity dataset as text.
purity_text = """run,purity_percent
1,99.1
2,98.7
3,99.0
4,97.9
5,99.2
6,98.1
"""

# Read text data using built-in csv tools.
reader = csv.DictReader(io.StringIO(purity_text))
purity_values = [float(row["purity_percent"]) for row in reader]

# Define specification and basic summary statistics.
spec_limit = 98.5
average_purity = stats.mean(purity_values)
minimum_purity = min(purity_values)
maximum_purity = max(purity_values)

# Compute spread and count specification failures.
purity_range = maximum_purity - minimum_purity
sample_stdev = stats.stdev(purity_values)
failed_count = sum(value < spec_limit for value in purity_values)

# Build an evidence-based quality conclusion.
if failed_count == 0 and sample_stdev <= 0.30:
    conclusion = "Accept: purity is stable and above specification."
elif average_purity >= spec_limit and failed_count > 0:
    conclusion = "Concern: average passes, but some runs fail specification."
else:
    conclusion = "Reject: summary indicates inadequate product quality."

# Print a compact engineering summary.
print("Solvent purity quality summary")
print(f"Average purity: {average_purity:.2f}%")
print(f"Minimum to maximum: {minimum_purity:.2f}% to {maximum_purity:.2f}%")
print(f"Range and stdev: {purity_range:.2f}%, {sample_stdev:.2f}%")
print(f"Runs below {spec_limit:.1f}% specification: {failed_count}")
print(conclusion)

```
