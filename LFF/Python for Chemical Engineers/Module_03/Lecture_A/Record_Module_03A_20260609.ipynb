## **1. Function Basics**

### **1.1. Function Definition**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780980498" width="250">



>* Functions package calculations for repeated engineering tasks
>* Reusable logic improves efficiency, consistency, and reviewability

>* Give each function one clear calculation purpose
>* Focused functions are easier to test and reuse

>* Separate fixed methods from changing case inputs
>* Build trusted functions for faster engineering analysis



```
#@title Python Code - Function Definition

# Functions make engineering calculations reusable.
# Parameters carry case specific information.
# Return values provide calculated engineering results.

# Define a focused residence time function.
def residence_time_minutes(volume_m3, flow_m3_per_min):
    tau_minutes = volume_m3 / flow_m3_per_min
    return tau_minutes

# Define a pressure conversion function.
def gauge_to_absolute_psia(gauge_psig, atmospheric_psia):
    absolute_psia = gauge_psig + atmospheric_psia
    return absolute_psia

# Store representative calculation inputs.
tank_volume_m3 = 12.0
feed_flow_m3_per_min = 3.0
line_pressure_psig = 35.0

# Use clear parameters when calling functions.
tau_result = residence_time_minutes(tank_volume_m3, feed_flow_m3_per_min)
absolute_result = gauge_to_absolute_psia(line_pressure_psig, 14.7)

# Print concise results for review.
print("Residence time:", tau_result, "min")
print("Absolute pressure:", absolute_result, "psia")
print("Same function can be reused with new inputs.")

# Reuse the residence time function again.
second_tau_result = residence_time_minutes(8.0, 2.0)
print("Second residence time:", second_tau_result, "min")

```

### **1.2. Pressure Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780980501" width="250">



>* Name pressure inputs by their exact role
>* Separate distinct pressures to avoid ambiguity

>* Distinguish absolute and gauge pressure
>* State pressure basis to prevent errors

>* Specify pressure units and conversions clearly
>* Name parameters by physical pressure role



```
#@title Python Code - Pressure Parameters

# Functions should name pressure meaning clearly.
# Absolute pressure often differs from gauge readings.
# Unit conversion makes reusable calculations safer.

# Define atmospheric pressure in kilopascals.
atm_pressure_kpa = 101.325

# Convert gauge pressure to absolute pressure.
def gauge_to_absolute_kpa(pressure_gauge_kpa, atmosphere_kpa=101.325):
    """Return absolute pressure in kPa from gauge pressure."""
    if pressure_gauge_kpa < -atmosphere_kpa:
        raise ValueError("Gauge pressure cannot be below vacuum.")

    return pressure_gauge_kpa + atmosphere_kpa

# Estimate ideal gas density using absolute pressure.
def ideal_gas_density(pressure_abs_kpa, temperature_k, molar_mass_kg_per_mol):
    """Return gas density in kg/m3 using absolute pressure."""
    if pressure_abs_kpa <= 0 or temperature_k <= 0:
        raise ValueError("Pressure and temperature must be positive.")

    gas_constant = 8.314462618
    pressure_abs_pa = pressure_abs_kpa * 1000.0

    density = pressure_abs_pa * molar_mass_kg_per_mol / (gas_constant * temperature_k)
    return density

# Use clear parameters for a nitrogen example.
n2_molar_mass = 0.0280134
reactor_pressure_gauge_kpa = 200.0
reactor_temperature_k = 300.0

# Convert pressure before calling the gas function.
reactor_pressure_abs_kpa = gauge_to_absolute_kpa(reactor_pressure_gauge_kpa, atm_pressure_kpa)
correct_density = ideal_gas_density(reactor_pressure_abs_kpa, reactor_temperature_k, n2_molar_mass)
wrong_density = ideal_gas_density(reactor_pressure_gauge_kpa, reactor_temperature_k, n2_molar_mass)

# Print compact results for comparison.
print(f"Gauge pressure: {reactor_pressure_gauge_kpa:.1f} kPa(g)")
print(f"Absolute pressure: {reactor_pressure_abs_kpa:.1f} kPa(abs)")
print(f"Correct nitrogen density: {correct_density:.3f} kg/m3")
print(f"If gauge used directly: {wrong_density:.3f} kg/m3")
print("Clear pressure parameters prevent hidden basis errors.")

```

### **1.3. Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780980500" width="250">



>* Return values are computed engineering results
>* They connect calculations to later analysis

>* Return clear values with meaningful units
>* Structure multiple results consistently

>* Return values make functions easier to test
>* Clear outputs support reliable reuse



```
#@title Python Code - Return Values

# Return values make engineering functions reusable.
# Functions should calculate, then return results.
# Returned values support testing and reuse.

def pressure_drop_metric(length_m, diameter_m, velocity_m_s, density_kg_m3, friction_factor):
    # Calculate Darcy Weisbach pressure drop.
    pressure_drop_pa = friction_factor * (length_m / diameter_m) * density_kg_m3 * velocity_m_s**2 / 2
    return pressure_drop_pa

# Define a heat duty function.
def heat_duty_metric(mass_flow_kg_s, heat_capacity_j_kg_k, temperature_change_k):
    heat_duty_w = mass_flow_kg_s * heat_capacity_j_kg_k * temperature_change_k
    return heat_duty_w

# Return two related engineering values.
def pipe_summary(length_m, diameter_m, velocity_m_s, density_kg_m3, friction_factor):
    pressure_drop_pa = pressure_drop_metric(length_m, diameter_m, velocity_m_s, density_kg_m3, friction_factor)
    pressure_drop_kpa = pressure_drop_pa / 1000
    return pressure_drop_pa, pressure_drop_kpa

# Use returned values in later calculations.
pipe_drop_pa = pressure_drop_metric(80, 0.05, 2.0, 998, 0.022)
heat_required_w = heat_duty_metric(1.5, 4180, 12)
raw_pa, clean_kpa = pipe_summary(80, 0.05, 2.0, 998, 0.022)

# Check results against simple expectations.
pipe_test_passed = abs(pipe_drop_pa - raw_pa) < 1e-9
heat_test_passed = heat_required_w > 0

# Print concise teaching results.
print("Pressure drop returned:", round(pipe_drop_pa, 1), "Pa")
print("Same pressure drop:", round(clean_kpa, 2), "kPa")
print("Heat duty returned:", round(heat_required_w / 1000, 2), "kW")
print("Returned pressure test passed:", pipe_test_passed)
print("Returned heat duty test passed:", heat_test_passed)

```

## **2. Engineering Interfaces**

### **2.1. Parameter Clarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780980511" width="250">



>* Clear parameters define required engineering inputs
>* Precise names prevent costly calculation mistakes

>* Choose parameters matching the engineering task
>* Use names that distinguish physical roles

>* State units, assumptions, and valid ranges
>* Prevent input errors and support reuse



```
#@title Python Code - Parameter Clarity

# Clear parameters prevent engineering calculation mistakes.
# Units belong directly in parameter names.
# Return values should communicate their meaning.

# Define a reusable heat duty function.
def heat_duty_kw(mass_flow_kg_s, heat_capacity_kj_kg_K, inlet_temp_C, outlet_temp_C):

    # Calculate the signed temperature change.
    delta_temp_K = outlet_temp_C - inlet_temp_C

    # Return heat duty using clear kilowatt units.
    return mass_flow_kg_s * heat_capacity_kj_kg_K * delta_temp_K

# Define a reusable pipe velocity function.
def pipe_velocity_m_s(volumetric_flow_m3_s, inside_diameter_m):

    # Reject impossible pipe diameters early.
    if inside_diameter_m <= 0:
        raise ValueError("inside_diameter_m must be positive")

    # Compute area from the inside diameter.
    pipe_area_m2 = 3.14159 * inside_diameter_m**2 / 4

    # Return average velocity with clear units.
    return volumetric_flow_m3_s / pipe_area_m2

# Run representative engineering calculation cases.
water_heater_kw = heat_duty_kw(1.8, 4.18, 25.0, 80.0)
hydrocarbon_cooler_kw = heat_duty_kw(2.5, 2.10, 120.0, 45.0)
pipe_velocity = pipe_velocity_m_s(0.012, 0.10)

# Print concise results for learner inspection.
print(f"Water heater duty: {water_heater_kw:.1f} kW")
print(f"Hydrocarbon cooler duty: {hydrocarbon_cooler_kw:.1f} kW")
print(f"Pipe velocity: {pipe_velocity:.2f} m/s")

# Show why vague parameter names are dangerous.
print("Clear names show mass flow, volume flow, temperature, and units.")

```

### **2.2. Input Output Clarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1780980513" width="250">



>* Make function inputs and calculations unambiguous
>* Return physically meaningful results to prevent errors

>* Choose inputs engineers can easily provide
>* Return outputs that support decisions

>* Clear interfaces make testing and teamwork easier
>* Explicit inputs prevent costly engineering mistakes



```
#@title Python Code - Input Output Clarity

# Clear interfaces prevent engineering unit mistakes.
# Function names should reveal physical meaning.
# Return values should be immediately interpretable.

# Define a reusable heat duty function.
def heat_duty_kw(mass_flow_kg_s, heat_capacity_kj_kg_k, delta_temp_k):
    """Return sensible heat duty in kilowatts."""

    if mass_flow_kg_s < 0 or heat_capacity_kj_kg_k < 0:
        raise ValueError("Flow and heat capacity must be nonnegative")

    heat_duty = mass_flow_kg_s * heat_capacity_kj_kg_k * delta_temp_k
    return heat_duty

# Define a reusable outlet temperature function.
def outlet_temp_c(inlet_temp_c, heat_duty_kw, mass_flow_kg_s, heat_capacity_kj_kg_k):
    """Return outlet temperature in degrees Celsius."""

    if mass_flow_kg_s <= 0 or heat_capacity_kj_kg_k <= 0:
        raise ValueError("Flow and heat capacity must be positive")

    delta_temp_c = heat_duty_kw / (mass_flow_kg_s * heat_capacity_kj_kg_k)
    outlet_temperature_c = inlet_temp_c + delta_temp_c

    return outlet_temperature_c

# Use named inputs for interface clarity.
water_flow_kg_s = 2.5
water_cp_kj_kg_k = 4.18
water_delta_temp_k = 12.0

# Calculate a physically named output.
duty_kw = heat_duty_kw(
    water_flow_kg_s, water_cp_kj_kg_k, water_delta_temp_k
)

# Reuse the duty in another clear function.
feed_inlet_temp_c = 25.0
feed_outlet_temp_c = outlet_temp_c(
    feed_inlet_temp_c, duty_kw, water_flow_kg_s, water_cp_kj_kg_k
)

# Print concise engineering results.
print("Function interface example: heat exchanger water stream")
print(f"Input mass flow: {water_flow_kg_s:.2f} kg/s")
print(f"Input heat capacity: {water_cp_kj_kg_k:.2f} kJ/(kg K)")
print(f"Returned heat duty: {duty_kw:.1f} kW")
print(f"Returned outlet temperature: {feed_outlet_temp_c:.1f} deg C")

```

### **2.3. Meaningful Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780980510" width="250">



>* Return results that answer the engineering question
>* Provide enough output for reuse

>* Name returns to show units and basis
>* Return engineering meaning, not just numbers

>* Balance useful detail with simple outputs
>* Return what the caller needs next



```
#@title Python Code - Meaningful Return Values

# Meaningful returns support reusable engineering calculations.
# Returned dictionaries can label units clearly.
# Tests compare representative heat exchanger cases.

# Define a function with clear input parameters.
def cooling_water_requirement(heat_kw, inlet_c, outlet_c):
    # Validate the temperature rise before division.
    if outlet_c <= inlet_c:
        raise ValueError("outlet_c must exceed inlet_c")

    # Use water heat capacity near room temperature.
    cp_kj_per_kg_c = 4.184
    delta_t_c = outlet_c - inlet_c
    flow_kg_per_s = heat_kw / (cp_kj_per_kg_c * delta_t_c)

    # Return engineering results with meaningful labels.
    result = {
        "heat_load_kw": heat_kw,
        "cooling_water_kg_s": flow_kg_per_s,
        "temperature_rise_c": delta_t_c,
    }

    return result

# Calculate one representative design case.
case_one = cooling_water_requirement(heat_kw=850, inlet_c=25, outlet_c=35)

# Calculate a smaller troubleshooting case.
case_two = cooling_water_requirement(heat_kw=120, inlet_c=20, outlet_c=28)

# Test the function with expected engineering behavior.
assert case_one["cooling_water_kg_s"] > case_two["cooling_water_kg_s"]
assert case_one["temperature_rise_c"] == 10

# Print compact results for interpretation.
print("Meaningful return values from reusable function")
print(f"Design heat load: {case_one['heat_load_kw']:.0f} kW")
print(f"Cooling water needed: {case_one['cooling_water_kg_s']:.2f} kg/s")
print(f"Temperature rise returned: {case_one['temperature_rise_c']:.1f} degC")
print(f"Troubleshooting flow: {case_two['cooling_water_kg_s']:.2f} kg/s")

```

## **3. Function Testing**

### **3.1. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780980505" width="250">



>* Test typical and known-answer cases
>* Include boundary cases to catch errors

>* Use realistic inputs and correct units
>* Check results against trusted sources

>* Tests preserve intended function behavior over time
>* Failures reveal changed assumptions or defects



```
#@title Python Code - Representative Test Cases

# Representative tests document engineering calculation behavior.
# Each case has inputs and expected results.
# Passing tests increase confidence in reusable functions.

# Define a reusable tank volume function.
def tank_volume_liters(diameter_m, height_m):

    """Return cylindrical tank volume in liters."""
    if diameter_m <= 0 or height_m <= 0:
        raise ValueError("dimensions must be positive")

    radius_m = diameter_m / 2.0
    volume_m3 = 3.141592653589793 * radius_m**2 * height_m
    return volume_m3 * 1000.0

# Store cases with independent expected answers.
test_cases = [
    ("typical lab tank", 1.0, 2.0, 1570.7963),
    ("known simple geometry", 2.0, 1.0, 3141.5927),

    ("small pilot vessel", 0.10, 0.50, 3.9270),
]

# Compare each calculation with tolerance.
passed_count = 0
for name, diameter, height, expected in test_cases:

    calculated = tank_volume_liters(diameter, height)
    error = abs(calculated - expected)
    passed = error < 0.01

    passed_count += int(passed)
    status = "PASS" if passed else "CHECK"
    print(f"{name}: {status}, {calculated:.2f} L")

# Test invalid engineering input separately.
try:
    tank_volume_liters(-1.0, 2.0)
    invalid_status = "CHECK"
except ValueError:
    invalid_status = "PASS"

print(f"invalid negative diameter: {invalid_status}")
summary_text = f"Summary: {passed_count} of {len(test_cases)} numeric tests passed."
print(summary_text)

```

### **3.2. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780980503" width="250">



>* Test realistic engineering inputs
>* Compare results with trusted calculations

>* Test normal, boundary, and changing conditions
>* Check results match obvious engineering trends

>* Tests show expected inputs and assumptions
>* Good cases reveal limits and prevent errors



```
#@title Python Code - Representative Test Cases

# Representative tests document reusable engineering function behavior.
# Each case checks one expected calculation situation.
# Small functions become safer with clear tests.

# Define a reusable heat duty function.
def heat_duty_kw(flow_kg_s, cp_kj_kgK, delta_t_K):
    """Return sensible heat duty in kilowatts."""
    if flow_kg_s < 0 or cp_kj_kgK < 0:
        raise ValueError("Flow and heat capacity must be nonnegative")

    if delta_t_K < 0:
        raise ValueError("Temperature change must be nonnegative")
    duty_kw = flow_kg_s * cp_kj_kgK * delta_t_K
    return duty_kw

# Store ordinary and boundary representative cases.
test_cases = [
    ("ordinary water heating", 2.0, 4.18, 15.0, 125.4),
    ("small positive flow", 0.05, 4.18, 10.0, 2.09),
    ("zero temperature change", 3.0, 4.18, 0.0, 0.0),
]

# Check numerical agreement using a small tolerance.
passed = 0
failed = 0
for name, flow, cp, dt, expected in test_cases:
    calculated = heat_duty_kw(flow, cp, dt)
    close_enough = abs(calculated - expected) < 1e-9
    if close_enough:
        passed = passed + 1
    else:
        failed = failed + 1

# Check one qualitative engineering behavior.
low_density_flow = heat_duty_kw(1.0, 4.18, 20.0)
high_density_flow = heat_duty_kw(2.0, 4.18, 20.0)
trend_ok = high_density_flow > low_density_flow

# Print a compact representative testing summary.
print("Representative heat-duty tests")
print(f"Passed numerical cases: {passed}")
print(f"Failed numerical cases: {failed}")
print(f"Increasing flow increases duty: {trend_ok}")
print("Boundary zero delta-T gives zero duty.")

```

### **3.3. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780980507" width="250">



>* Test realistic and boundary engineering cases
>* Check both calculations and engineering meaning

>* Use independently checkable expected results
>* Choose realistic cases with acceptable tolerances

>* Test normal, boundary, and problematic inputs
>* Document reasons to support trusted reuse



```
#@title Python Code - Representative Test Cases

# Representative tests build confidence in functions.
# Expected values should be independently checked.
# Edge cases reveal hidden calculation problems.

# Define a reusable heat duty function.
def heat_duty_kw(mass_flow_kg_s, cp_kj_kg_k, delta_t_k):
    """Return sensible heat duty in kilowatts."""

    # Reject physically impossible negative inputs.
    if mass_flow_kg_s < 0 or cp_kj_kg_k < 0:
        raise ValueError("Mass flow and heat capacity must be nonnegative")

    # Calculate kilowatts from consistent metric units.
    duty_kw = mass_flow_kg_s * cp_kj_kg_k * delta_t_k
    return duty_kw

# Store representative engineering test cases.
test_cases = [
    ("moderate water heating", 2.0, 4.18, 10.0, 83.6),
    ("zero temperature change", 3.0, 4.18, 0.0, 0.0),
    ("cooling case negative duty", 1.5, 2.50, -8.0, -30.0),
]

# Compare each result with its expected value.
passed_tests = 0
for name, flow, cp, dt, expected in test_cases:
    result = heat_duty_kw(flow, cp, dt)
    passed = abs(result - expected) < 1e-9

    # Count successful representative cases.
    if passed:
        passed_tests += 1

    # Print one concise result line.
    status = "PASS" if passed else "FAIL"
    print(f"{name}: {result:.1f} kW, expected {expected:.1f} -> {status}")

# Test a deliberately invalid input.
try:
    heat_duty_kw(-1.0, 4.18, 5.0)
    invalid_status = "FAIL"
except ValueError:
    invalid_status = "PASS"

# Summarize the complete test set.
total_tests = len(test_cases) + 1
passed_tests = passed_tests + (invalid_status == "PASS")
print(f"Invalid input check -> {invalid_status}")
print(f"Passed {passed_tests} of {total_tests} representative tests")

```
