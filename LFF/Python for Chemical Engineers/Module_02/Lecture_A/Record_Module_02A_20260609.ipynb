## **1. Ordered Chemical Data**

### **1.1. Temperature Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780978562" width="250">



>* Temperature lists store readings in process order
>* Changing order misrepresents the process data

>* Group related temperature readings in order
>* Reveal trends, gradients, and unusual behavior

>* Keep temperature units consistent
>* Know what each position represents



```
#@title Python Code - Temperature Lists

# Temperature lists preserve ordered process measurements.
# Each position represents a meaningful condition.
# Built-in functions summarize small sequences safely.

# Create reactor temperatures in Celsius.
temperatures_c = [24.8, 36.5, 51.2, 68.9, 80.1]

# Store matching times in minutes.
times_min = [0, 5, 10, 15, 20]

# Confirm both ordered lists align.
valid_sequence = len(temperatures_c) == len(times_min)

# Access the first and last readings.
start_temp = temperatures_c[0]
final_temp = temperatures_c[-1]

# Use slicing for the heating period.
mid_run_temps = temperatures_c[1:4]

# Summarize the temperature list.
lowest_temp = min(temperatures_c)
highest_temp = max(temperatures_c)
average_temp = sum(temperatures_c) / len(temperatures_c)

# Print concise teaching results.
print("Aligned time and temperature lists:", valid_sequence)
print("Temperature list, degrees Celsius:", temperatures_c)
print("Start and final temperatures:", start_temp, final_temp)
print("Middle run temperatures:", mid_run_temps)
print("Minimum, maximum, average:", lowest_temp, highest_temp, round(average_temp, 1))
print("Ordered pairs show process history:", list(zip(times_min, temperatures_c)))

```

### **1.2. Creating Ordered Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780978564" width="250">



>* Preserve measurement order to keep meaning
>* Use lists and tuples to organize data

>* Use lists for data that changes
>* Use tuples for fixed data records

>* Keep related measurements aligned by order
>* Include meaning, units, and context



```
#@title Python Code - Creating Ordered Sequences

# Lists store measurements that may change.
# Tuples store records that should stay fixed.
# Order preserves engineering meaning in sequences.

# Create a list for changing reactor temperatures.
reactor_temperatures_c = [82.1, 84.0, 85.5, 86.2]

# Add a new measurement to the ordered list.
reactor_temperatures_c.append(87.0)

# Correct the first reading after calibration review.
reactor_temperatures_c[0] = 82.4

# Create a tuple for fixed mixture component order.
mixture_components = ("ethanol", "water")

# Store specimen dimensions as a stable tuple.
specimen_dimensions_cm = (2.5, 10.0, 30.0)

# Print the ordered data containers.
print("Temperature list:", reactor_temperatures_c)
print("Component tuple:", mixture_components)
print("Dimension tuple cm:", specimen_dimensions_cm)

# Show that first and last positions matter.
first_temperature = reactor_temperatures_c[0]
last_temperature = reactor_temperatures_c[-1]

# Summarize the ordered temperature sequence briefly.
print("First temperature C:", first_temperature)
print("Last temperature C:", last_temperature)
print("Number of readings:", len(reactor_temperatures_c))

# Compare list flexibility with tuple stability.
print("Lists can be updated; tuples represent fixed order.")

```

### **1.3. Chemical Data Containers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780978566" width="250">



>* Containers group related engineering measurements
>* Order preserves sequence context for analysis

>* Lists store changing ordered measurements together
>* Treat related readings as one analyzable collection

>* Tuples store fixed ordered engineering records
>* Choose containers to communicate data purpose



```
#@title Python Code - Chemical Data Containers

# Lists hold measurements that may change.
# Tuples hold records intended to stay fixed.
# Both preserve order in chemical data.

# Store reactor temperatures in an editable list.
reactor_temperatures_c = [84.5, 85.1, 86.0, 85.7]

# Add a new measurement to the list.
reactor_temperatures_c.append(86.3)

# Store fixed pipe geometry in a tuple.
pipe_geometry_in = (2.0, 12.0, 120.0)

# Unpack tuple values into clear names.
diameter_in, wall_thickness_in, length_ft = pipe_geometry_in

# Summarize the ordered temperature readings.
count_readings = len(reactor_temperatures_c)
average_temperature_c = sum(reactor_temperatures_c) / count_readings

# Access first and last ordered readings.
first_temperature_c = reactor_temperatures_c[0]
last_temperature_c = reactor_temperatures_c[-1]

# Print compact results for inspection.
print("Temperature list:", reactor_temperatures_c)
print("Fixed pipe tuple:", pipe_geometry_in)
print("First and last temperatures:", first_temperature_c, last_temperature_c)
print("Average temperature, C:", round(average_temperature_c, 2))
print("Pipe diameter in, length ft:", diameter_in, length_ft)

```

## **2. Sequence Access**

### **2.1. Indexing Elements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780978551" width="250">



>* Indexes locate specific ordered sequence elements
>* Python indexing starts at zero

>* Use negative indexes to access final elements
>* Check sequence length to avoid indexing errors

>* Update lists when measurements need correction
>* Use tuples for fixed, stable records



```
#@title Python Code - Indexing Elements

# Practice indexing ordered engineering sequences.
# Lists can be read or changed.
# Tuples can be read safely.

# Store hourly reactor temperatures in Celsius.
temperatures_c = [72.4, 75.1, 78.6, 80.2]

# Store fixed sampling port names as a tuple.
ports = ("feed", "middle", "product")

# Check sequence lengths before indexing elements.
if len(temperatures_c) >= 4 and len(ports) >= 3:
    first_temperature = temperatures_c[0]
    final_temperature = temperatures_c[-1]

    middle_port = ports[1]
    final_port = ports[-1]

    # Correct one list value after calibration review.
    temperatures_c[1] = 75.4

    # Use selected elements in a simple calculation.
    temperature_change = temperatures_c[-1] - temperatures_c[0]

    print("First temperature:", first_temperature, "deg C")
    print("Final temperature:", final_temperature, "deg C")
    print("Middle sampling port:", middle_port)
    print("Final sampling port:", final_port)
    print("Corrected second temperature:", temperatures_c[1], "deg C")
    print("Overall temperature rise:", round(temperature_change, 2), "deg C")

# Finish with a compact sequence length check.
print("Temperature count:", len(temperatures_c))

```

### **2.2. Slicing Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780978553" width="250">



>* Select meaningful ranges from ordered data
>* Preserve order for easier analysis

>* Start included, stop excluded in slices
>* End slices preserve ordered measurement meaning

>* Use slices to prepare data for analysis
>* Focus on meaningful process segments



```
#@title Python Code - Slicing Sequences

# Demonstrate slicing ordered reactor measurements.
# Use small lists for clear output.
# Compare process intervals without loops.

# Store minute labels and temperatures.
minutes = [0, 1, 2, 3, 4, 5]
temps_c = [22.0, 28.5, 35.0, 41.0, 46.5, 50.0]

# Slice startup and later operation.
startup_temps = temps_c[0:3]
later_temps = temps_c[3:6]

# Slice from the end safely.
recent_temps = temps_c[-2:]
paired_recent = list(zip(minutes[-2:], recent_temps))

# Transform one slice with looping.
startup_f = [temp * 9 / 5 + 32 for temp in startup_temps]
recent_average = sum(recent_temps) / len(recent_temps)

# Print compact engineering results.
print("All temperatures C:", temps_c)
print("Startup slice C:", startup_temps)
print("Later slice C:", later_temps)
print("Recent minute-temperature pairs:", paired_recent)
print("Startup slice F:", startup_f)
print("Recent average C:", round(recent_average, 2))

```

### **2.3. Looping Through Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780978555" width="250">



>* Visit each measurement in order
>* Apply checks or conversions consistently

>* Loop using values or positions as needed
>* Indexes add context when reporting unusual readings

>* Transform measurements while preserving order
>* Protect data integrity with careful loops



```
#@title Python Code - Looping Through Sequences

# Loop through ordered reactor temperature data.
# Track both measurement values and positions.
# Store transformed values without changing originals.

# Store hourly reactor temperatures in Celsius.
temperatures_c = [84.5, 88.0, 91.2, 89.7]

# Define the safe operating temperature limit.
safe_limit_c = 90.0

# Prepare an empty list for Fahrenheit values.
temperatures_f = []

# Loop through values and their hourly indexes.
for hour_index, temp_c in enumerate(temperatures_c):

    # Convert each Celsius reading to Fahrenheit.
    temp_f = temp_c * 9 / 5 + 32

    # Append converted values in matching order.
    temperatures_f.append(round(temp_f, 1))

    # Report only readings above the limit.
    if temp_c > safe_limit_c:

        # Print the hour and unsafe temperature.
        print("High reading at hour", hour_index, ":", temp_c, "C")

# Summarize the transformed sequence briefly.
print("Original Celsius readings:", temperatures_c)

# Show converted Fahrenheit readings.
print("Converted Fahrenheit readings:", temperatures_f)

# Use built-ins to summarize all readings.
print("Maximum Celsius reading:", max(temperatures_c))

```

## **3. Sequence Summaries**

### **3.1. Sequence Sums**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780978558" width="250">



>* Sum ordered measurements stored in sequences
>* Totals support quick engineering calculations

>* Sum only values that meaningfully accumulate
>* Use engineering judgment before reporting totals

>* Built-in sums reduce calculation mistakes
>* Reusable totals support engineering decisions



```
#@title Python Code - Sequence Sums

# Sequence sums summarize ordered engineering measurements.
# Built-in sum reduces manual arithmetic mistakes.
# Engineers decide whether totals are meaningful.

# Store solvent volumes from cleaning intervals.
solvent_liters = [12.5, 10.0, 11.2, 9.8]

# Store catalyst charges from separate reactors.
catalyst_kg = (3.0, 2.5, 2.8, 3.2)

# Check sequences contain matching usable measurements.
valid_lengths = len(solvent_liters) > 0 and len(catalyst_kg) > 0

# Calculate meaningful accumulated totals using sum.
total_solvent = sum(solvent_liters)
total_catalyst = sum(catalyst_kg)

# Calculate an average from temperature sums.
reactor_temps_c = [348.0, 351.5, 349.2, 350.1]
temp_average = sum(reactor_temps_c) / len(reactor_temps_c)

# Display concise engineering summaries.
print("Sequence Sums for Chemical Engineering")
print(f"Measurements available: {valid_lengths}")
print(f"Total solvent used: {total_solvent:.1f} L")
print(f"Total catalyst charged: {total_catalyst:.1f} kg")
print(f"Average reactor temperature: {temp_average:.1f} °C")
print("Temperature sum supports averaging, not final reporting.")

```

### **3.2. Mean Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780978560" width="250">



>* Means summarize typical measurement levels
>* Built-ins simplify totals and counts

>* Relate means to engineering meaning
>* Use ordered sequences for context

>* Means can hide spikes or faulty values
>* Check data quality and interpret carefully



```
#@title Python Code - Mean Measurements

# Store ordered reactor measurements.
# Summarize values using built-ins.
# Interpret the mean carefully.

# No package installation is needed.

# Store temperatures as a list.
temperatures_c = [82.4, 83.1, 82.8, 84.0, 83.3]

# Store flow measurements as a tuple.
flows_gpm = (14.8, 15.1, 14.9, 15.0, 15.2)

# Check sequence lengths before division.
temp_count = len(temperatures_c)

flow_count = len(flows_gpm)

# Compute totals with built-in sum.
temp_total = sum(temperatures_c)

flow_total = sum(flows_gpm)

# Compute mean measurements safely.
mean_temp = temp_total / temp_count

mean_flow = flow_total / flow_count

# Compare means with target conditions.
target_temp = 83.0

target_flow = 15.0

# Print compact engineering summaries.
print("Mean reactor temperature:", round(mean_temp, 2), "deg C")

print("Temperature offset from target:", round(mean_temp - target_temp, 2), "deg C")

print("Mean cooling-water flow:", round(mean_flow, 2), "gal/min")

print("Flow offset from target:", round(mean_flow - target_flow, 2), "gal/min")

print("Measurements summarized:", temp_count + flow_count, "total values")

```

### **3.3. Average Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780978557" width="250">



>* Averages summarize repeated process measurements clearly
>* Compute using total and measurement count

>* Average comparable measurements with equal importance
>* Use engineering judgment for meaningful summaries

>* Averages can hide important process variation
>* Compare averages with other sequence summaries



```
#@title Python Code - Average Measurements

# Lists can store repeated process measurements.
# Averages summarize typical operating behavior.
# Built-ins make simple summaries readable.

# Store reactor temperatures in a list.
reactor_temps_c = [82.1, 83.0, 82.7, 83.4, 82.8]

# Store pipeline pressures in a tuple.
pipeline_pressures_psi = (48.5, 49.1, 48.8, 49.0)

# Count measurements before dividing totals.
temp_count = len(reactor_temps_c)
pressure_count = len(pipeline_pressures_psi)

# Compute averages using sum and length.
avg_temp_c = sum(reactor_temps_c) / temp_count
avg_pressure_psi = sum(pipeline_pressures_psi) / pressure_count

# Build short summary text safely.
temp_summary = f"Average reactor temperature: {avg_temp_c:.2f} °C"
pressure_summary = f"Average pipeline pressure: {avg_pressure_psi:.2f} psi"

# Print compact engineering summaries.
print("Sequence Summaries: Average Measurements")
print(temp_summary)
print(pressure_summary)
print(f"Temperature readings counted: {temp_count}")
print(f"Pressure readings counted: {pressure_count}")

```
