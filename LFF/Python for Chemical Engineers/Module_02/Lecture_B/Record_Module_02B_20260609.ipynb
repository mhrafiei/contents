## **1. Engineering Dictionaries**

### **1.1. Property Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780979467" width="250">



>* Labels give engineering values clear meaning
>* Dictionaries make property data easier to read

>* Clear labels prevent property confusion
>* Dictionaries organize meaningful engineering records

>* Choose clear, specific property labels
>* Labels make engineering data self-describing



```
#@title Python Code - Property Labels

# Dictionaries connect engineering labels with values.
# Clear labels reduce property confusion.
# This example stores stream properties.

# Define one labeled process stream record.
feed_stream = {
    "stream name": "Reactor feed",
    "temperature_C": 85.0,
    "pressure_kPa": 240.0,

    "mass_flow_kg_per_h": 1250.0,
    "liquid_density_kg_per_m3": 920.0,
}

# Read values by their engineering labels.
name = feed_stream["stream name"]
temperature = feed_stream["temperature_C"]
pressure = feed_stream["pressure_kPa"]

# Add a new labeled property safely.
feed_stream["pipe_diameter_in"] = 3.0
label_count = len(feed_stream)

# Print a compact engineering summary.
print("Dictionary record:", name)
print("Temperature label gives", temperature, "C")
print("Pressure label gives", pressure, "kPa")
print("Available property labels:", label_count)

# Show labels that explain each value.
for label, value in feed_stream.items():
    print(label, "=", value)

# Confirm the dictionary supports labeled lookup.
print("Mass flow is", feed_stream["mass_flow_kg_per_h"], "kg/h")

```

### **1.2. Property Lookup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780979465" width="250">



>* Dictionaries retrieve values using property names
>* Named lookup reduces order-based mistakes

>* Find properties by meaningful engineering labels
>* Labels keep large process records understandable

>* Label lookups stay stable as records change
>* Named values improve clarity and traceability



```
#@title Python Code - Property Lookup

# Dictionaries connect property labels to values.
# Engineers retrieve data by meaningful names.
# This example uses a stream record.

# Store labeled process properties together.
feed_stream = {
    "temperature_C": 85.0,
    "pressure_kPa": 220.0,
    "mass_flow_kg_s": 1.75,
}

# Look up properties directly by name.
temperature = feed_stream["temperature_C"]
pressure = feed_stream["pressure_kPa"]
flow_rate = feed_stream["mass_flow_kg_s"]

# Use looked-up values in calculations.
pressure_bar = pressure / 100.0
residence_factor = flow_rate / pressure_bar

# Print concise engineering results.
print("Feed stream property lookup")
print(f"Temperature: {temperature:.1f} C")
print(f"Pressure: {pressure:.2f} kPa")
print(f"Pressure: {pressure_bar:.2f} bar")
print(f"Mass flow: {flow_rate:.2f} kg/s")

# Check whether an optional property exists.
property_name = "viscosity_mPa_s"
lookup_value = feed_stream.get(property_name, "not recorded")

# Show safe lookup for missing data.
print(f"Viscosity: {lookup_value}")
print(f"Flow per bar: {residence_factor:.3f} kg/s/bar")

```

### **1.3. Labeled Property Lookup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780979469" width="250">



>* Retrieve values using meaningful property names
>* Match data access to engineering language

>* Labels reduce property lookup mistakes
>* Names clarify varied engineering records

>* Labels improve team communication and shared understanding
>* Clear labels reduce ambiguity and analysis errors



```
#@title Python Code - Labeled Property Lookup

# Dictionaries connect names with engineering values.
# Labels reduce position based lookup mistakes.
# This example retrieves ethanol properties clearly.

# Store ethanol properties using descriptive labels.
ethanol_properties = {
    "molecular_weight_g_per_mol": 46.07,
    "boiling_point_C": 78.37,

    "liquid_density_kg_per_m3": 789,
    "heat_of_vaporization_kJ_per_mol": 38.56,
}

# Lookup values by property labels.
density = ethanol_properties["liquid_density_kg_per_m3"]
boiling_point = ethanol_properties["boiling_point_C"]

# Use get for optional properties.
viscosity = ethanol_properties.get("viscosity_mPa_s", "not recorded")

# Show readable engineering lookup results.
print("Ethanol property lookup")
print(f"Density: {density} kg/m^3")
print(f"Boiling point: {boiling_point} deg C")
print(f"Viscosity: {viscosity}")

# Combine labels for a process stream record.
feed_stream = {
    "stream_name": "ethanol feed",
    "temperature_C": 25.0,

    "pressure_kPa": 101.3,
    "total_flow_kmol_per_h": 12.5,
}

# Retrieve stream values by engineering meaning.
label = "total_flow_kmol_per_h"
flow_rate = feed_stream[label]

# Print one labeled stream result.
print(f"{feed_stream['stream_name']} flow: {flow_rate} kmol/h")

```

## **2. Text File Handling**

### **2.1. Opening Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780979462" width="250">



>* Open files to access stored engineering data
>* Choose modes for reading or writing

>* Choose modes for reading, writing, or appending
>* Protect data and keep outputs traceable

>* Use correct file paths and names
>* Close files to save and release resources



```
#@title Python Code - Opening Files

# Text files connect programs with stored engineering data.
# Opening modes control reading, writing, and appending safely.
# Context managers close files after operations finish.

# Define a small process data filename.
file_name = "feed_temperatures.txt"

# Create a tiny text file using write mode.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write("Feed stream temperatures, deg C\n")
    file_object.write("Run 1: 72.5\n")

    file_object.write("Run 2: 74.1\n")
    file_object.write("Run 3: 73.4\n")

# Reopen the same file using read mode.
with open(file_name, "r", encoding="utf-8") as file_object:
    file_contents = file_object.read()

# Append one new observation without erasing earlier data.
with open(file_name, "a", encoding="utf-8") as file_object:
    file_object.write("Run 4: 75.0\n")

# Read the updated file as separate lines.
with open(file_name, "r", encoding="utf-8") as file_object:
    updated_lines = file_object.readlines()

# Print concise results from each file operation.
print("Created file:", file_name)
print("Initial text preview:", file_contents.splitlines()[0])
print("Number of updated lines:", len(updated_lines))
print("Last stored measurement:", updated_lines[-1].strip())

```

### **2.2. Reading Text Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780979463" width="250">



>* Text files bring engineering data into programs
>* File layout guides correct data interpretation

>* Read files line by line
>* Split, convert, and store values

>* Handle messy files and missing values
>* Keep metadata for reliable engineering analysis



```
#@title Python Code - Reading Text Files

# Read small chemical engineering text files.
# Convert characters into useful records.
# Store labeled measurements using dictionaries.

# Create a tiny instrument export file.
file_name = "reactor_temperatures.txt"

# Each data line stores one observation.
lines = [
    "time_min,temperature_C,pressure_kPa\n",
    "0,22.5,101.3\n",

    "5,35.8,118.9\n",
    "10,48.2,136.4\n",
]

# Write text exactly like an instrument might.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.writelines(lines)

# Prepare an empty list for records.
records = []

# Open the file for reading text.
with open(file_name, "r", encoding="utf-8") as file_object:
    header = file_object.readline().strip().split(",")

    # Read remaining lines one observation at time.
    for raw_line in file_object:
        parts = raw_line.strip().split(",")

        # Convert text fields into numeric values.
        record = {
            header[0]: float(parts[0]),
            header[1]: float(parts[1]),

            header[2]: float(parts[2]),
        }
        records.append(record)

# Check that data was actually loaded.
if len(records) == 0:
    raise ValueError("No reactor measurements were loaded.")

# Calculate a simple engineering summary.
average_temp = sum(item["temperature_C"] for item in records) / len(records)

# Show only compact readable results.
print(f"Loaded records: {len(records)}")
print(f"Column names: {header}")
print(f"First record dictionary: {records[0]}")
print(f"Average temperature: {average_temp:.1f} °C")

```

### **2.3. Writing Text Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780979460" width="250">



>* Save engineering data in readable text files
>* Include labels, units, and clear context

>* Use consistent structure and separators
>* Make files easy to analyze later

>* Choose overwrite or append based on purpose
>* Record units, precision, and close files



```
#@title Python Code - Writing Text Files

# Text files preserve engineering calculations.
# Labels and units prevent ambiguity.
# Context managers close files safely.

# Store small heat exchanger measurements.
measurements = [
    {"time_min": 0, "temp_c": 22.0, "pressure_kpa": 101.3},
    {"time_min": 5, "temp_c": 36.8, "pressure_kpa": 104.9},

    {"time_min": 10, "temp_c": 48.5, "pressure_kpa": 108.2},
]

# Choose a clear output filename.
file_name = "heat_exchanger_log.txt"

# Write labels, units, and data rows.
with open(file_name, "w", encoding="utf-8") as file:
    file.write("Heat exchanger trial log\n")
    file.write("time_min,temp_C,pressure_kPa\n")

    for row in measurements:
        line = (
            f"{row['time_min']},"
            f"{row['temp_c']:.1f},"

            f"{row['pressure_kpa']:.1f}\n"
        )
        file.write(line)

# Read back a few lines for inspection.
with open(file_name, "r", encoding="utf-8") as file:
    saved_lines = file.readlines()

# Show only the compact saved content.
print("Created file:", file_name)
print("Saved text preview:")
print("".join(saved_lines[:5]))

```

## **3. Structured Records**

### **3.1. Nested Process Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780979458" width="250">



>* Group related process data in nested containers
>* Keep equipment properties clearly connected

>* Group repeated equipment data as complete records
>* Labeled values make process data clearer

>* Keep real process context together
>* Design clearer, easier-to-maintain records



```
#@title Python Code - Nested Process Records

# Nested records keep process data organized.
# Inner dictionaries describe meaningful engineering groups.
# Lists can store repeated structured records.

# Create one process unit record.
reactor_record = {
    "tag": "R-101",

    "status": "running",
    "conditions": {
        "temperature_C": 82.0,

        "pressure_kPa": 240.0,
        "volume_m3": 3.5,
    },

    "feed_stream": {
        "name": "aqueous feed",
        "flow_kg_h": 1200.0,

        "composition": {
            "water": 0.72,
            "ethanol": 0.28,
        },

    },
    "outlet_streams": [
        {
            "name": "product",

            "flow_kg_h": 1185.0,
            "temperature_C": 84.0,
        },

        {
            "name": "vent",
            "flow_kg_h": 15.0,
            "temperature_C": 40.0,

        },
    ],
}

# Read values using nested keys.
tag_value = reactor_record["tag"]
feed_flow = reactor_record["feed_stream"]["flow_kg_h"]
product_flow = reactor_record["outlet_streams"][0]["flow_kg_h"]

# Update one nested operating value.
reactor_record["conditions"]["temperature_C"] = 85.0
new_temperature = reactor_record["conditions"]["temperature_C"]

# Check one simple mass balance.
outlet_total = sum(
    stream["flow_kg_h"]
    for stream in reactor_record["outlet_streams"]
)

# Print a compact engineering summary.
print("Nested process record summary")
print(f"Unit tag: {tag_value}")
print(f"Updated temperature: {new_temperature} C")
print(f"Feed flow: {feed_flow} kg/h")
print(f"Total outlet flow: {outlet_total} kg/h")
print(f"Product flow: {product_flow} kg/h")
print(f"Mass balance error: {feed_flow - outlet_total} kg/h")

```

### **3.2. Nested Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780979454" width="250">



>* Group related stream data inside records
>* Use inner records for meaningful subcategories

>* Group related data to avoid confusing labels
>* Match record structure to engineering systems

>* Nested records organize data across engineering tools
>* Grouped details make records expandable and meaningful



```
#@title Python Code - Nested Records

# Nested records group engineering details.
# Dictionaries can hold other dictionaries.
# This example models process streams.

# Create one nested stream record.
feed_stream = {
    "name": "Feed gas",
    "conditions": {"temperature_C": 35.0, "pressure_kPa": 250.0},

    "flow": {"value": 120.0, "units": "kmol/h"},
    "composition": {"methane": 0.82, "ethane": 0.10},

    "quality": {"source": "lab sample", "status": "checked"},
}

# Add another component safely.
feed_stream["composition"]["carbon_dioxide"] = 0.08

# Validate composition before using it.
fractions = feed_stream["composition"].values()
composition_sum = sum(fractions)

# Store another record inside a list.
product_stream = {
    "name": "Product gas",
    "conditions": {"temperature_C": 28.0, "pressure_kPa": 210.0},

    "flow": {"value": 112.0, "units": "kmol/h"},
    "composition": {"methane": 0.88, "ethane": 0.09},

    "quality": {"source": "simulator", "status": "estimated"},
}

# Combine records into a structured dataset.
streams = [feed_stream, product_stream]

# Print selected nested values clearly.
print("Stream count:", len(streams))
print("Feed pressure kPa:", feed_stream["conditions"]["pressure_kPa"])
print("Feed CO2 mole fraction:", feed_stream["composition"]["carbon_dioxide"])
print("Feed composition sum:", round(composition_sum, 3))
print("Product data source:", product_stream["quality"]["source"])
print("First stream flow units:", streams[0]["flow"]["units"])

```

### **3.3. Structured Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Chemical Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780979456" width="250">



>* Samples store many labeled engineering details
>* Nested containers keep related measurements together

>* Organized records preserve sample labels
>* Structure prevents mistakes and supports analysis

>* Organize trials as collections of sample records
>* Build flexible models with combined containers



```
#@title Python Code - Structured Samples

# Build nested records for chemical samples.
# Keep labels beside every measured value.
# Compare structured samples without losing context.

# Create two structured engineering sample records.
sample_001 = {
    "id": "WW-001",
    "location": "influent channel",

    "conditions": {
        "temperature_C": 22.4,
        "flow_m3_h": 180.0,

    },
    "composition_mg_L": {
        "COD": 520.0,
        "TSS": 210.0,

    },
    "replicate_pH": [
        7.05,
        7.10,
        7.08,

    ],
    "flags": [
        "field meter calibrated",
    ],
}

# Store another sample with the same structure.
sample_002 = {
    "id": "WW-002",
    "location": "final effluent",

    "conditions": {
        "temperature_C": 21.7,
        "flow_m3_h": 172.0,

    },
    "composition_mg_L": {
        "COD": 38.0,
        "TSS": 12.0,

    },
    "replicate_pH": [
        7.42,
        7.39,
        7.41,

    ],
    "flags": [
        "lab result reviewed",
    ],
}

# Combine records into one trial container.
trial_samples = [
    sample_001,
    sample_002,
]

# Compute each average pH from replicates.
for sample in trial_samples:
    ph_values = sample["replicate_pH"]

    average_ph = sum(ph_values) / len(ph_values)
    sample["average_pH"] = round(average_ph, 2)

# Estimate treatment removal using labeled fields.
inlet_cod = trial_samples[0]["composition_mg_L"]["COD"]
outlet_cod = trial_samples[1]["composition_mg_L"]["COD"]

cod_removal = 100.0 * (inlet_cod - outlet_cod) / inlet_cod
cod_removal = round(cod_removal, 1)

# Print compact readable summaries for learners.
print("Structured sample summaries:")
for sample in trial_samples:
    sample_id = sample["id"]

    sample_location = sample["location"]
    sample_cod = sample["composition_mg_L"]["COD"]

    sample_ph = sample["average_pH"]
    print(sample_id, sample_location, sample_cod, "mg/L COD", sample_ph, "pH")

# Show one process-level calculation.
print("COD removal across plant:", cod_removal, "%")

```
