## **1. Probability Classifier**

### **1.1. Linear Score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_01_01.jpg?v=1783617180" width="250">



>* Features combine into one evidence score
>* Score precedes probability or final decision

>* Features add weighted evidence to one score
>* Baseline and weights make decisions interpretable

>* Raw scores are not probabilities
>* Sigmoid converts evidence into decision probabilities



```
#@title Python Code - Linear Score

# Logistic regression starts with evidence.
# Features receive weights before probabilities.
# Civil examples make scores concrete.

import numpy as np
import matplotlib.pyplot as plt

# Store three bridge inspection cases.
case_names = np.array(["A", "B", "C"])
crack_mm = np.array([1.0, 4.0, 8.0])
traffic_kips = np.array([20.0, 35.0, 55.0])

# Combine features into one matrix.
features = np.column_stack((crack_mm, traffic_kips))
assert features.shape == (3, 2)

# Choose illustrative learned weights.
weights = np.array([0.55, 0.04])
bias = -3.0
assert weights.shape[0] == features.shape[1]

# Compute the raw linear scores.
scores = features @ weights + bias
probabilities = 1 / (1 + np.exp(-scores))

# Print compact teaching results.
print("Linear score = bias + weighted features")
for name, score, prob in zip(case_names, scores, probabilities):
    print(f"Bridge {name}: score {score:.2f}, probability {prob:.2f}")

# Plot scores before probability conversion.
plt.figure(figsize=(6, 3.5))
plt.bar(case_names, scores, color="steelblue")
plt.axhline(0, color="black", linewidth=1)
plt.title("Raw Linear Scores for Bridge Risk")
plt.xlabel("Bridge inspection case")
plt.ylabel("Linear score, not probability")
plt.tight_layout()
plt.show()

```

### **1.2. Sigmoid Probability**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_01_02.jpg?v=1783617182" width="250">



>* Linear scores are not probabilities yet
>* Sigmoid converts scores into class probabilities

>* Extreme scores give near-certain probabilities
>* Middle scores show model uncertainty

>* Probabilities show closeness to decision boundaries
>* Scores compare confidence before final decisions



```
#@title Python Code - Sigmoid Probability

# Logistic regression maps scores into probabilities.
# Sigmoid keeps outputs between zero and one.
# Civil decisions can use probability thresholds.

import math
import numpy as np
import matplotlib.pyplot as plt

# Define a stable sigmoid function.
def sigmoid(score_value):
    clipped_score = np.clip(score_value, -50, 50)
    return 1 / (1 + np.exp(-clipped_score))

# Create simple bridge inspection features.
crack_width_mm = np.array([0.2, 0.8, 1.5, 2.4])
traffic_million = np.array([0.5, 1.2, 2.0, 3.1])

# Combine features into linear scores.
intercept_value = -3.0
crack_weight = 1.4
traffic_weight = 0.7

linear_scores = (
    intercept_value
    + crack_weight * crack_width_mm
    + traffic_weight * traffic_million
)

# Convert scores into class probabilities.
failure_probabilities = sigmoid(linear_scores)
threshold_value = 0.50
predicted_classes = failure_probabilities >= threshold_value

# Print a compact teaching table.
print("Sigmoid probability example for bridge risk")
print("crack_mm traffic_M score probability decision")
for index_value in range(len(crack_width_mm)):
    decision_text = "inspect" if predicted_classes[index_value] else "monitor"
    print(f"{crack_width_mm[index_value]:.1f} {traffic_million[index_value]:.1f} {linear_scores[index_value]:.2f} {failure_probabilities[index_value]:.2f} {decision_text}")

# Plot sigmoid curve and example scores.
score_grid = np.linspace(-6, 6, 200)
probability_grid = sigmoid(score_grid)
plt.figure(figsize=(7, 4))

plt.plot(score_grid, probability_grid, label="sigmoid probability")
plt.scatter(linear_scores, failure_probabilities, color="red", label="bridge examples")
plt.axhline(threshold_value, color="gray", linestyle="--", label="0.50 threshold")

plt.xlabel("linear score from engineering features")
plt.ylabel("estimated positive-class probability")
plt.title("Sigmoid Converts Scores Into Probabilities")
plt.legend()

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **1.3. Binary Class Probability**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_01_03.jpg?v=1783617184" width="250">



>* Two-class outcomes use positive class probabilities
>* Scores become confidence estimates, not hard decisions

>* Probabilities show confidence beyond class labels
>* Thresholds convert probabilities into practical decisions

>* Positive and negative probabilities sum to one
>* Probabilities are estimates, not guaranteed truths



```
#@title Python Code - Binary Class Probability

# Logistic regression turns scores into probabilities.
# Civil features can support binary decisions.
# Thresholds convert probabilities into class labels.

import math
import numpy as np
import matplotlib.pyplot as plt

# Store small civil engineering inspection examples.
crack_mm = np.array([0.2, 0.8, 1.5, 2.4, 3.2])
load_ratio = np.array([0.35, 0.50, 0.65, 0.80, 0.95])
labels = np.array([0, 0, 1, 1, 1])

# Validate matching feature and label lengths.
assert crack_mm.shape == load_ratio.shape == labels.shape
features = np.column_stack([crack_mm, load_ratio])
weights = np.array([1.4, 3.0])

# Combine features into one linear score.
bias = -4.0
scores = features @ weights + bias
probabilities = 1 / (1 + np.exp(-scores))

# Convert positive probability into negative probability.
negative_probabilities = 1 - probabilities
threshold = 0.50
predicted_labels = (probabilities >= threshold).astype(int)

# Print compact probability interpretation table.
print("Binary class probability example")
print("case | P(repair) | P(no repair) | decision")
for index, probability in enumerate(probabilities):
    decision = "repair" if predicted_labels[index] else "monitor"
    print(f"{index+1:>4} | {probability:>9.2f} | {negative_probabilities[index]:>12.2f} | {decision}")

# Show how sigmoid maps scores into probabilities.
score_grid = np.linspace(-6, 6, 200)
probability_grid = 1 / (1 + np.exp(-score_grid))
plt.figure(figsize=(6, 4))

# Plot the smooth probability curve.
plt.plot(score_grid, probability_grid, label="sigmoid probability")
plt.axhline(threshold, color="gray", linestyle="--", label="threshold")
plt.scatter(scores, probabilities, color="red", label="inspection cases")

# Label axes for engineering interpretation.
plt.xlabel("linear engineering score")
plt.ylabel("probability of repair class")
plt.title("Logistic Regression Binary Class Probability")
plt.legend()

# Display exactly one teaching plot.
plt.tight_layout()
plt.show()

```

## **2. Manual Model Training**

### **2.1. Sigmoid Prediction**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_02_01.jpg?v=1783617194" width="250">



>* Features combine into one raw score
>* Sigmoid converts scores into class probabilities

>* Sigmoid shows gradual uncertainty between classes
>* Thresholds turn probabilities into final actions

>* Features combine into a raw evidence score
>* Prediction errors guide loss and updates



```
#@title Python Code - Sigmoid Prediction

# Manual logistic regression starts with scores.
# Sigmoid converts scores into probabilities.
# Civil examples can use inspection features.

import math
import numpy as np
import matplotlib.pyplot as plt

# Store tiny bridge inspection feature data.
crack_mm = np.array([0.2, 1.0, 2.5, 4.0])
traffic_kips = np.array([12.0, 18.0, 25.0, 35.0])

# Combine features using chosen demonstration weights.
weight_crack = 1.10
weight_traffic = 0.08
bias_term = -5.00

# Validate matching feature vector lengths.
assert crack_mm.shape == traffic_kips.shape
assert crack_mm.size == 4

# Define sigmoid using the math library.
def sigmoid(raw_score):
    return 1.0 / (1.0 + math.exp(-raw_score))

# Compute raw scores and sigmoid probabilities.
raw_scores = bias_term + weight_crack * crack_mm + weight_traffic * traffic_kips
probabilities = np.array([sigmoid(score) for score in raw_scores])

# Convert probabilities into maintenance decisions.
threshold = 0.50
decisions = probabilities >= threshold

# Print compact teaching results only.
print("Sigmoid prediction for bridge maintenance risk")
print(f"Weights: crack={weight_crack}, traffic={weight_traffic}, bias={bias_term}")
print(f"Threshold for action: {threshold}")

# Show each example without printing large arrays.
for index, probability in enumerate(probabilities):
    label = "inspect" if decisions[index] else "monitor"
    print(f"Bridge {index+1}: score={raw_scores[index]:.2f}, probability={probability:.2f}, decision={label}")

# Plot sigmoid curve and bridge scores.
x_values = np.linspace(-8.0, 8.0, 200)
y_values = 1.0 / (1.0 + np.exp(-x_values))

# Create one clear probability plot.
plt.figure(figsize=(7, 4))
plt.plot(x_values, y_values, label="sigmoid curve")
plt.scatter(raw_scores, probabilities, color="red", label="bridges")

# Add threshold reference line.
plt.axhline(threshold, color="gray", linestyle="--", label="threshold")
plt.xlabel("Raw model score")
plt.ylabel("Predicted probability")
plt.title("Sigmoid Prediction From Raw Scores")

# Display the single teaching plot.
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Gradient Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_02_02.jpg?v=1783617198" width="250">



>* Compare predictions with actual outcomes
>* Adjust feature weights gradually through practice

>* Updates move weights toward lower prediction error
>* Learning rate controls step size and stability

>* Feature weights change based on prediction errors
>* Updates build better decision rules over time



```
#@title Python Code - Gradient Updates

# Manual logistic training uses small gradient steps.
# Civil features become probabilities through sigmoid prediction.
# Updates reduce mistakes across repeated examples.

# Import numerical tools and plotting support.
import numpy as np
import matplotlib.pyplot as plt

# Set deterministic formatting for readable output.
np.set_printoptions(precision=3, suppress=True)

# Create tiny bridge inspection training data.
X_raw = np.array([
    [12.0, 1.0],
    [25.0, 2.0],

    [40.0, 4.0],
    [55.0, 6.0],
    [70.0, 8.0],
    [85.0, 9.0],
])

# Labels mark whether urgent repair is needed.
y = np.array([0, 0, 0, 1, 1, 1], dtype=float)

# Validate the feature matrix and labels.
assert X_raw.shape == (6, 2)
assert y.shape[0] == X_raw.shape[0]

# Standardize features for stable gradient updates.
means = X_raw.mean(axis=0)
scales = X_raw.std(axis=0)
X = (X_raw - means) / scales

# Add an intercept column for bias learning.
ones = np.ones((X.shape[0], 1))
X_design = np.column_stack((ones, X))

# Define sigmoid conversion from score to probability.
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# Define average binary cross entropy loss.
def binary_loss(probabilities, targets):
    eps = 1e-9
    clipped = np.clip(probabilities, eps, 1.0 - eps)

    losses = -(targets * np.log(clipped) + (1 - targets) * np.log(1 - clipped))
    return float(np.mean(losses))

# Initialize weights and learning settings.
weights = np.zeros(X_design.shape[1])
learning_rate = 0.4
loss_history = []

# Train with quiet batch gradient descent.
for step in range(80):
    probabilities = sigmoid(X_design @ weights)
    gradient = X_design.T @ (probabilities - y) / y.size

    weights = weights - learning_rate * gradient
    loss_history.append(binary_loss(probabilities, y))

# Compute final probabilities and decisions.
final_probabilities = sigmoid(X_design @ weights)
threshold = 0.50
predictions = (final_probabilities >= threshold).astype(int)

# Summarize learning without printing large arrays.
print("Manual logistic regression gradient updates")
print(f"Initial loss: {loss_history[0]:.3f}")
print(f"Final loss: {loss_history[-1]:.3f}")
print(f"Learned weights: {weights.round(3)}")
print(f"Threshold used: {threshold:.2f}")
print(f"Predicted classes: {predictions.tolist()}")
print(f"True classes:      {y.astype(int).tolist()}")

# Plot loss decrease across update steps.
plt.figure(figsize=(6, 4))
plt.plot(loss_history, color="steelblue", linewidth=2)
plt.xlabel("Gradient update step")
plt.ylabel("Binary loss")
plt.title("Loss decreases during manual training")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Loss Intuition**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_02_03.jpg?v=1783617196" width="250">



>* Loss measures prediction confidence errors
>* Confident wrong predictions receive stronger penalties

>* Loss rewards well-calibrated class probabilities
>* Examples adjust weights toward better separation

>* Loss guides repeated prediction updates
>* Falling loss shows better label alignment



```
#@title Python Code - Loss Intuition

# Manual training reveals loss intuition.
# Confident wrong probabilities are costly.
# Small updates reduce future mistakes.

import math
import numpy as np
import matplotlib.pyplot as plt

# Store tiny civil engineering inspection examples.
crack_mm = np.array([1.0, 2.0, 3.5, 5.0, 7.0, 9.0])
labels = np.array([0, 0, 0, 1, 1, 1])
assert crack_mm.shape == labels.shape

# Scale crack width for stable updates.
x_mean = crack_mm.mean()
x_std = crack_mm.std()
x_scaled = (crack_mm - x_mean) / x_std

# Define sigmoid probability prediction.
def sigmoid(score):
    clipped = np.clip(score, -30, 30)
    return 1 / (1 + np.exp(-clipped))

# Define binary cross entropy loss.
def binary_loss(y_true, y_prob):
    eps = 1e-9
    safe_prob = np.clip(y_prob, eps, 1 - eps)
    return -np.mean(y_true * np.log(safe_prob) + (1 - y_true) * np.log(1 - safe_prob))

# Start with simple manual weights.
weight = 0.0
bias = 0.0
learning_rate = 0.8
loss_history = []

# Train silently using gradient updates.
for step in range(25):
    probabilities = sigmoid(weight * x_scaled + bias)
    loss_history.append(binary_loss(labels, probabilities))
    error = probabilities - labels
    weight -= learning_rate * np.mean(error * x_scaled)
    bias -= learning_rate * np.mean(error)

# Compare loss for different confidence levels.
loss_good = binary_loss(np.array([1]), np.array([0.99]))
loss_bad = binary_loss(np.array([1]), np.array([0.01]))
print(f"Loss when confidently correct: {loss_good:.3f}")
print(f"Loss when confidently wrong: {loss_bad:.3f}")
print(f"Initial training loss: {loss_history[0]:.3f}")
print(f"Final training loss: {loss_history[-1]:.3f}")

# Evaluate decisions using a chosen threshold.
final_prob = sigmoid(weight * x_scaled + bias)
threshold = 0.50
predicted = (final_prob >= threshold).astype(int)
accuracy = np.mean(predicted == labels)
print(f"Threshold decision accuracy: {accuracy:.2f}")

# Plot loss decrease during manual training.
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(loss_history) + 1), loss_history, marker="o")
plt.xlabel("Gradient update step")
plt.ylabel("Binary loss")
plt.title("Loss decreases as probabilities improve")
plt.grid(True, alpha=0.3)
plt.show()

```

## **3. Threshold Decisions**

### **3.1. Decision Thresholds**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_03_01.jpg?v=1783617186" width="250">



>* Logistic regression outputs probabilities, not labels
>* Thresholds turn probabilities into real decisions

>* Thresholds balance missed cases and false alarms
>* Lower cutoffs act sooner, but less selectively

>* Thresholds change decisions, not model probabilities
>* Choose thresholds using context-specific performance costs



```
#@title Python Code - Decision Thresholds

# Thresholds convert probabilities into engineering decisions.
# Different cutoffs create different mistake patterns.
# This example compares three simple thresholds.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Store predicted failure probabilities for bridge segments.
probabilities = np.array([0.08, 0.18, 0.31, 0.44, 0.52, 0.67, 0.79, 0.91])

# Store observed inspection outcomes for each segment.
actual_failures = np.array([0, 0, 0, 1, 0, 1, 1, 1])

# Validate matching lengths before comparing decisions.
assert probabilities.shape == actual_failures.shape, "Arrays must match."

# Define thresholds representing different risk tolerances.
thresholds = np.array([0.30, 0.50, 0.70])

# Count decision outcomes for one threshold.
def summarize_threshold(cutoff, probabilities, actual_failures):
    predicted_failures = (probabilities >= cutoff).astype(int)
    true_positive = int(np.sum((predicted_failures == 1) & (actual_failures == 1)))
    false_positive = int(np.sum((predicted_failures == 1) & (actual_failures == 0)))

    false_negative = int(np.sum((predicted_failures == 0) & (actual_failures == 1)))
    true_negative = int(np.sum((predicted_failures == 0) & (actual_failures == 0)))
    accuracy = float(np.mean(predicted_failures == actual_failures))

    return [cutoff, true_positive, false_positive, false_negative, true_negative, accuracy]

# Build a compact table for threshold comparison.
rows = [summarize_threshold(t, probabilities, actual_failures) for t in thresholds]

# Name columns using common classification terms.
columns = ["threshold", "TP", "FP", "FN", "TN", "accuracy"]

# Create a small readable results table.
results = pd.DataFrame(rows, columns=columns)

# Print only a compact summary table.
print("Bridge segment failure decisions by threshold:")
print(results.to_string(index=False, float_format=lambda x: f"{x:.2f}"))

# Plot probabilities and threshold lines.
plt.figure(figsize=(7, 4))
plt.scatter(range(len(probabilities)), probabilities, c=actual_failures, cmap="coolwarm", s=80)
plt.axhline(0.30, color="green", linestyle="--", label="threshold 0.30")

plt.axhline(0.50, color="black", linestyle="--", label="threshold 0.50")
plt.axhline(0.70, color="purple", linestyle="--", label="threshold 0.70")
plt.xlabel("Bridge segment index")
plt.ylabel("Predicted failure probability")

plt.title("Changing thresholds changes decisions, not probabilities")
plt.ylim(0, 1)
plt.legend(loc="lower right")
plt.show()

```

### **3.2. Threshold Selection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_03_02.jpg?v=1783617188" width="250">



>* Thresholds turn probabilities into class decisions
>* Best cutoff depends on mistake consequences

>* Compare false positive and false negative costs
>* Choose thresholds for real-world consequences

>* Test cutoffs on validation data
>* Choose thresholds for real decision tradeoffs



```
#@title Python Code - Threshold Selection

# Thresholds turn probabilities into engineering decisions.
# Different cutoffs create different mistake patterns.
# Validation data helps choose practical thresholds.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Store predicted failure probabilities and true outcomes.
probabilities = np.array([0.08, 0.18, 0.27, 0.36, 0.44, 0.52, 0.61, 0.73, 0.82, 0.91])
actual_failures = np.array([0, 0, 0, 1, 0, 1, 1, 1, 1, 1])
thresholds = np.array([0.30, 0.50, 0.70])

# Check that both arrays describe identical projects.
assert probabilities.shape == actual_failures.shape
assert probabilities.size == 10

# Evaluate each threshold using simple counts.
rows = []
for cutoff in thresholds:
    predicted_failures = (probabilities >= cutoff).astype(int)
    true_positive = int(((predicted_failures == 1) & (actual_failures == 1)).sum())
    false_positive = int(((predicted_failures == 1) & (actual_failures == 0)).sum())

    false_negative = int(((predicted_failures == 0) & (actual_failures == 1)).sum())
    true_negative = int(((predicted_failures == 0) & (actual_failures == 0)).sum())
    accuracy = (true_positive + true_negative) / probabilities.size
    rows.append([cutoff, true_positive, false_positive, false_negative, accuracy])

# Summarize threshold tradeoffs in a compact table.
summary = pd.DataFrame(rows, columns=["threshold", "TP", "FP", "FN", "accuracy"])
print("Threshold selection for bridge inspection alerts")
print(summary.to_string(index=False))

# Plot probabilities and threshold decision lines.
plt.figure(figsize=(7, 4))
plt.scatter(range(1, 11), probabilities, c=actual_failures, cmap="coolwarm", s=80)
for cutoff in thresholds:
    plt.axhline(cutoff, linestyle="--", linewidth=1, label=f"threshold {cutoff:.2f}")

# Label the plot for decision interpretation.
plt.xlabel("Inspection case number")
plt.ylabel("Predicted probability of repair need")
plt.title("Changing thresholds changes repair alerts")
plt.legend(loc="lower right")
plt.show()

```

### **3.3. Confusion Matrix**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_05/Lecture_A/image_03_03.jpg?v=1783617190" width="250">



>* Summarizes threshold-based classification outcomes
>* Shows true, false, positive, negative decisions

>* Error costs depend on the application
>* Judge mistakes by real-world consequences

>* Thresholds shift positives, negatives, and errors
>* Confusion matrices guide practical decision tradeoffs



```
#@title Python Code - Confusion Matrix

# Confusion matrices summarize threshold based decisions.
# Civil risk examples make errors concrete.
# Small arrays keep calculations transparent.

import numpy as np
import matplotlib.pyplot as plt

# Store actual bridge inspection outcomes.
y_true = np.array([1, 0, 1, 0, 1, 0, 0, 1])

# Store logistic regression risk probabilities.
y_prob = np.array([0.82, 0.35, 0.61, 0.55, 0.48, 0.22, 0.40, 0.73])

# Choose the probability threshold.
threshold = 0.50

# Convert probabilities into class decisions.
y_pred = (y_prob >= threshold).astype(int)

# Validate matching vector lengths.
assert y_true.shape == y_pred.shape

# Count the four confusion matrix outcomes.
tp = int(np.sum((y_true == 1) & (y_pred == 1)))
tn = int(np.sum((y_true == 0) & (y_pred == 0)))
fp = int(np.sum((y_true == 0) & (y_pred == 1)))
fn = int(np.sum((y_true == 1) & (y_pred == 0)))

# Arrange counts in standard matrix form.
confusion = np.array([[tn, fp], [fn, tp]])

# Compute simple evaluation measures.
accuracy = (tp + tn) / len(y_true)
precision = tp / max(tp + fp, 1)
recall = tp / max(tp + fn, 1)

# Print a compact decision summary.
print(f"Threshold: {threshold:.2f}")
print(f"TN={tn}, FP={fp}, FN={fn}, TP={tp}")
print(f"Accuracy={accuracy:.2f}, Precision={precision:.2f}, Recall={recall:.2f}")

# Plot the confusion matrix.
fig, ax = plt.subplots(figsize=(4, 3))
im = ax.imshow(confusion, cmap="Blues")

# Label rows and columns clearly.
ax.set_xticks([0, 1], labels=["Pred 0", "Pred 1"])
ax.set_yticks([0, 1], labels=["Actual 0", "Actual 1"])
ax.set_title("Confusion Matrix at Threshold 0.50")

# Write counts inside each cell.
for row in range(2):
    for col in range(2):
        ax.text(col, row, confusion[row, col], ha="center", va="center")

# Add color scale and display.
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()

```
