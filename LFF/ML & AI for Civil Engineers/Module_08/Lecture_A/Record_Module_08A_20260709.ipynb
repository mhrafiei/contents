## **1. Image Arrays**

### **1.1. Pixel Grid Basics**

>* Images are grids of tiny pixels
>* Pixel values become analyzable numerical arrays

>* Pixel positions preserve real-world spatial relationships
>* Grid patterns reveal defects and features

>* Different images store different pixel values
>* Arrays enable measurement, filtering, and automation



```
#@title Python Code - Pixel Grid Basics

# Pixel grids store visual information numerically.
# Rows and columns preserve spatial locations.
# Small arrays can represent inspection images.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny grayscale pavement grid.
pavement_grid = np.array([
    [180, 178, 176, 174, 172, 170],
    [181, 179,  40,  38, 171, 169],

    [182, 180,  42,  39, 170, 168],
    [183, 181, 179, 177, 169, 167],
    [184, 182, 180, 178, 168, 166],
    [185, 183, 181, 179, 167, 165],
])

# Validate the grid shape before plotting.
rows, columns = pavement_grid.shape
assert rows == 6 and columns == 6

# Select one pixel representing a dark crack.
crack_row, crack_column = 1, 2
crack_value = pavement_grid[crack_row, crack_column]

# Print a compact summary for learners.
print(f"Grid shape: {rows} rows by {columns} columns")
print(f"Pixel at row {crack_row}, column {crack_column}: {crack_value}")
print("Lower values appear darker in this grayscale image.")

# Display the pixel grid as an image.
fig, ax = plt.subplots(figsize=(5, 5))
image = ax.imshow(pavement_grid, cmap="gray", vmin=0, vmax=255)

# Add grid lines to show pixel boundaries.
ax.set_xticks(np.arange(-0.5, columns, 1), minor=True)
ax.set_yticks(np.arange(-0.5, rows, 1), minor=True)
ax.grid(which="minor", color="yellow", linewidth=1)

# Label rows and columns like array indices.
ax.set_xticks(np.arange(columns))
ax.set_yticks(np.arange(rows))
ax.set_xlabel("Column index")
ax.set_ylabel("Row index")

# Highlight the selected dark pixel.
ax.scatter(crack_column, crack_row, s=180, facecolors="none", edgecolors="red")
ax.set_title("Tiny pavement image as a NumPy pixel grid")
fig.colorbar(image, ax=ax, label="Brightness value")

# Show the single teaching plot.
plt.show()

```

```
#@title Python Code - Pixel Grid Basics

# Pixel grids store visual engineering information.
# NumPy arrays make pixels measurable.
# Small examples reveal image structure.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny grayscale bridge-deck image.
deck = np.array([
    [210, 210, 205, 200, 198, 196],
    [212, 208, 202, 195, 190, 188],

    [215, 210,  40,  35, 185, 182],
    [218, 212,  45,  38, 180, 178],

    [220, 216, 210, 205, 175, 170],
    [222, 218, 214, 208, 172, 168]
], dtype=np.uint8)

# Validate the grid before visualization.
assert deck.shape == (6, 6)
assert deck.dtype == np.uint8

# Print compact facts about the array.
print("Grid shape:", deck.shape)
print("Pixel at row 2, column 2:", deck[2, 2])
print("Dark crack-like pixels:", int(np.sum(deck < 60)))

# Display the pixel grid as an image.
fig, ax = plt.subplots(figsize=(5, 5))
image = ax.imshow(deck, cmap="gray", vmin=0, vmax=255)

# Add row and column tick labels.
ax.set_xticks(np.arange(deck.shape[1]))
ax.set_yticks(np.arange(deck.shape[0]))

# Label the civil engineering interpretation.
ax.set_title("Tiny grayscale bridge-deck pixel grid")
ax.set_xlabel("Column position")
ax.set_ylabel("Row position")

# Annotate each pixel with brightness.
for row in range(deck.shape[0]):
    for col in range(deck.shape[1]):
        value = int(deck[row, col])
        ax.text(col, row, value, ha="center", va="center")

# Add a colorbar for brightness values.
fig.colorbar(image, ax=ax, label="Brightness, 0 dark to 255 light")
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Pixel Grid Basics

# Images become grids of numbers.
# Rows and columns locate pixels.
# Brightness values describe visual intensity.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny grayscale bridge-deck patch.
image_grid = np.array([
    [220, 220, 215, 210, 205, 200],
    [218, 216, 210, 205, 198, 195],
    [215, 210,  40,  35, 190, 188],
    [212, 205,  45,  38, 185, 182],
    [210, 202, 195, 188, 180, 176],
    [205, 198, 190, 182, 175, 170]
], dtype=np.uint8)

# Validate the grid before visualization.
assert image_grid.ndim == 2
assert image_grid.size > 0

# Store simple physical scale information.
pixel_size_cm = 2.5
rows, columns = image_grid.shape

# Print compact facts about the image array.
print(f"Array shape: {rows} rows by {columns} columns")
print(f"Pixel value range: {image_grid.min()} to {image_grid.max()}")
print(f"Each pixel represents about {pixel_size_cm} centimeters")
print("Dark pixels may represent cracks or shadows")

# Plot the pixel grid as grayscale data.
fig, ax = plt.subplots(figsize=(5, 5))
image_plot = ax.imshow(image_grid, cmap="gray", vmin=0, vmax=255)

# Label rows and columns clearly.
ax.set_title("Tiny grayscale pixel grid")
ax.set_xlabel("Column index")
ax.set_ylabel("Row index")

# Show every pixel index on axes.
ax.set_xticks(np.arange(columns))
ax.set_yticks(np.arange(rows))
ax.grid(color="yellow", linewidth=1)

# Add pixel values inside cells.
for row in range(rows):
    for col in range(columns):
        value = image_grid[row, col]
        text_color = "white" if value < 100 else "black"
        ax.text(col, row, str(value), ha="center", va="center", color=text_color)

# Add a compact color scale.
fig.colorbar(image_plot, ax=ax, label="Brightness intensity")
plt.tight_layout()
plt.show()

```

## **2. CNN Building Blocks**

### **2.1. Convolution Filters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_02_01.jpg?v=1783625659" width="250">



>* Filters scan images for local visual patterns
>* Small observations build broader image understanding

>* Same filter finds features anywhere
>* Feature maps show detection locations

>* Filters learn layered visual patterns from data
>* Outputs need validation and human judgment



```
#@title Python Code - Convolution Filters

# Convolution filters detect local image patterns.
# Small kernels slide across pixel grids.
# Feature maps show stronger pattern responses.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny bridge-like grayscale image.
image = np.zeros((12, 16), dtype=float)
image[2:4, 2:14] = 0.75
image[7:9, 1:15] = 0.55
image[4:10, [4, 8, 12]] = 0.95

# Add a dark crack-like diagonal feature.
for row_index in range(5, 10):
    col_index = row_index + 1
    image[row_index, col_index] = 0.05

# Define a vertical edge detection filter.
vertical_filter = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1],
], dtype=float)

# Define a simple manual convolution function.
def apply_filter(pixel_grid, kernel):
    grid_rows, grid_cols = pixel_grid.shape
    kernel_rows, kernel_cols = kernel.shape

    assert kernel_rows == 3 and kernel_cols == 3
    output = np.zeros((grid_rows - 2, grid_cols - 2))

    for row_index in range(grid_rows - 2):
        for col_index in range(grid_cols - 2):
            patch = pixel_grid[row_index:row_index + 3, col_index:col_index + 3]
            output[row_index, col_index] = np.sum(patch * kernel)

    return output

# Apply the filter to create a feature map.
feature_map = apply_filter(image, vertical_filter)
feature_strength = np.abs(feature_map)

# Summarize the small convolution result.
print("Original image shape:", image.shape)
print("Filter shape:", vertical_filter.shape)
print("Feature map shape:", feature_map.shape)
print("Strongest edge response:", round(float(feature_strength.max()), 2))

# Visualize the image, filter, and response.
fig, axes = plt.subplots(1, 3, figsize=(11, 3))
axes[0].imshow(image, cmap="gray", vmin=0, vmax=1)
axes[0].set_title("Civil image grid")
axes[1].imshow(vertical_filter, cmap="coolwarm")

axes[1].set_title("Vertical edge filter")
axes[2].imshow(feature_strength, cmap="magma")
axes[2].set_title("Feature map strength")

# Remove axes for cleaner teaching visuals.
for axis in axes:
    axis.set_xticks([])
    axis.set_yticks([])

plt.tight_layout()
plt.show()

```

### **2.2. Pooling Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_02_02.jpg?v=1783625653" width="250">



>* Pooling summarizes feature maps into smaller regions
>* Max or average pooling keeps key evidence

>* Handles small image shifts more reliably
>* Shrinks feature maps for faster learning

>* Pooling builds larger patterns from small features
>* Use carefully when spatial precision matters



```
#@title Python Code - Pooling Layers

# Pooling summarizes local feature responses.
# Small shifts become less important.
# Civil images often need robust summaries.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny feature map example.
feature_map = np.array([
    [0, 1, 0, 2, 1, 0],
    [1, 8, 2, 1, 0, 1],

    [0, 2, 1, 0, 7, 2],
    [1, 0, 1, 2, 1, 0],
    [0, 1, 6, 1, 0, 1],

    [2, 0, 1, 0, 1, 5]
], dtype=float)

# Validate the pooling window compatibility.
height, width = feature_map.shape
assert height % 2 == 0 and width % 2 == 0

# Define simple two by two pooling.
def pool_two_by_two(array, mode):
    output = np.zeros((array.shape[0] // 2, array.shape[1] // 2))

    for row in range(0, array.shape[0], 2):
        for col in range(0, array.shape[1], 2):
            patch = array[row:row + 2, col:col + 2]

            if mode == "max":
                output[row // 2, col // 2] = np.max(patch)
            else:
                output[row // 2, col // 2] = np.mean(patch)

    return output

# Compute max and average pooling summaries.
max_pooled = pool_two_by_two(feature_map, "max")
avg_pooled = pool_two_by_two(feature_map, "average")

# Print only compact teaching summaries.
print("Original feature map shape:", feature_map.shape)
print("Max pooled shape:", max_pooled.shape)
print("Average pooled shape:", avg_pooled.shape)
print("Max pooling keeps strongest local evidence.")
print("Average pooling creates smoother local summaries.")

# Prepare one figure with three panels.
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
images = [feature_map, max_pooled, avg_pooled]

titles = ["Feature map", "Max pooling", "Average pooling"]
for axis, image, title in zip(axes, images, titles):
    shown = axis.imshow(image, cmap="viridis", vmin=0, vmax=8)
    axis.set_title(title)

    axis.set_xticks([])
    axis.set_yticks([])

# Add one shared colorbar for activation strength.
fig.colorbar(shown, ax=axes, shrink=0.75, label="activation")
plt.tight_layout()
plt.show()

```

### **2.3. Dense Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_02_03.jpg?v=1783625656" width="250">



>* Combine earlier visual features into interpretations
>* Support judgments about cracks, stains, or wear

>* Fully connected units combine extracted features
>* Late layers turn patterns into classifications

>* Many connections can cause overfitting risks
>* Predictions still need engineering judgment



```
#@title Python Code - Dense Layers

# Dense layers combine extracted visual evidence.
# This example uses tiny manual features.
# No machine learning library is needed.

import numpy as np
import matplotlib.pyplot as plt

# Create three simplified bridge-deck feature strengths.
feature_names = ["dark lines", "rough texture", "color change"]
features = np.array([0.85, 0.60, 0.30], dtype=float)

# Validate the feature vector before dense calculations.
assert features.shape == (3,), "Expected exactly three features."
assert np.all((features >= 0) & (features <= 1))

# Each dense unit connects to every input feature.
weights = np.array([[1.2, 0.7, 0.2], [0.2, 0.5, 1.1]])
bias = np.array([-0.4, -0.2], dtype=float)

# Validate dense layer dimensions before matrix multiplication.
assert weights.shape == (2, 3), "Expected two dense units."
assert bias.shape == (2,), "Expected two bias values."

# Compute dense scores using all feature connections.
scores = weights @ features + bias
labels = ["possible crack", "surface staining"]

# Convert scores into simple probabilities for interpretation.
shifted = scores - np.max(scores)
probabilities = np.exp(shifted) / np.sum(np.exp(shifted))

# Print a compact summary for learners.
print("Dense layer input features:", dict(zip(feature_names, features.round(2))))
print("Dense layer scores:", dict(zip(labels, scores.round(2))))
print("Predicted class:", labels[int(np.argmax(probabilities))])

# Visualize how dense units weight every feature.
fig, ax = plt.subplots(figsize=(7, 3.5))
image = ax.imshow(weights, cmap="coolwarm", vmin=-1.2, vmax=1.2)

# Label rows and columns for interpretation.
ax.set_xticks(np.arange(len(feature_names)))
ax.set_xticklabels(feature_names, rotation=20, ha="right")
ax.set_yticks(np.arange(len(labels)))
ax.set_yticklabels(labels)

# Add readable weight values inside cells.
for row in range(weights.shape[0]):
    for col in range(weights.shape[1]):
        ax.text(col, row, f"{weights[row, col]:.1f}", ha="center", va="center")

# Finish the single teaching plot.
ax.set_title("Dense layer: every output uses every feature")
fig.colorbar(image, ax=ax, label="connection weight")
plt.tight_layout()
plt.show()

```

## **3. Simple CNN Architectures**

### **3.1. Manual Filter Operations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_03_01.jpg?v=1783625662" width="250">



>* Filters scan pixel grids for local patterns
>* Strong responses highlight edges, cracks, textures

>* Scan images to find edges and patterns
>* Blur, sharpen, and transform pixel grids

>* Plots reveal where filters detect patterns
>* Interpret results carefully; noise can mislead



```
#@title Python Code - Manual Filter Operations

# Manual filters reveal local image patterns.
# Tiny arrays can represent civil images.
# One plot compares several filter effects.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create a small synthetic pavement image.
image_size = 40
image = np.full((image_size, image_size), 0.65)

# Add a darker vertical crack feature.
image[:, 19:21] = 0.15
image[10:30, 24:26] = 0.25

# Add a bright painted lane marking.
image[28:32, 4:36] = 0.95
image = np.clip(image, 0.0, 1.0)

# Define simple three by three filters.
vertical_edge = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
horizontal_edge = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
blur_filter = np.ones((3, 3)) / 9.0

# Apply one filter using manual convolution.
def apply_filter(gray_image, kernel):
    rows, cols = gray_image.shape
    assert kernel.shape == (3, 3)

    output = np.zeros((rows - 2, cols - 2))
    for row in range(1, rows - 1):
        for col in range(1, cols - 1):
            patch = gray_image[row - 1:row + 2, col - 1:col + 2]

            output[row - 1, col - 1] = np.sum(patch * kernel)
    return output

# Compute filtered activation style images.
vertical_result = apply_filter(image, vertical_edge)
horizontal_result = apply_filter(image, horizontal_edge)
blur_result = apply_filter(image, blur_filter)

# Print a short interpretation summary.
print("Synthetic pavement grid shape:", image.shape)
print("Vertical filter highlights left-right intensity changes.")
print("Horizontal filter highlights top-bottom intensity changes.")
print("Blur filter averages nearby pixels and reduces sharpness.")

# Create exactly one figure with four panels.
fig, axes = plt.subplots(1, 4, figsize=(13, 3))
titles = ["Original", "Vertical edges", "Horizontal edges", "Blurred"]
images = [image, vertical_result, horizontal_result, blur_result]

# Display each image-like array consistently.
for axis, title, shown in zip(axes, titles, images):
    axis.imshow(shown, cmap="gray")
    axis.set_title(title)
    axis.axis("off")

# Add a compact lecture style title.
fig.suptitle("Manual filter operations on a pavement-like pixel grid")
plt.tight_layout()
plt.show()

```

### **3.2. Activation Maps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_03_02.jpg?v=1783625664" width="250">



>* Shows where filters respond strongly
>* Reveals civil features after filtering

>* Plots reveal where filters respond strongly
>* Interpret bright activations with context and caution

>* Activation maps connect manual filters and CNNs
>* Visualizations reveal patterns without full explanations



```
#@title Python Code - Activation Maps

# Activation maps show filter responses.
# Synthetic pixels represent civil imagery.
# Manual convolution builds visual intuition.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a small synthetic pavement image.
image = np.full((24, 24), 0.35, dtype=float)
image[:, 11:13] = 0.85

# Add a horizontal lane marking.
image[16:18, 3:21] = 0.75
image[5:20, 6] = 0.10

# Define simple edge detecting filters.
vertical_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
horizontal_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

# Validate image and filter dimensions.
assert image.ndim == 2 and vertical_filter.shape == (3, 3)
assert horizontal_filter.shape == (3, 3) and min(image.shape) >= 3

# Apply one manual valid convolution.
def convolve_valid(pixel_grid, kernel):
    rows = pixel_grid.shape[0] - kernel.shape[0] + 1
    cols = pixel_grid.shape[1] - kernel.shape[1] + 1

    activation = np.zeros((rows, cols), dtype=float)
    for row in range(rows):
        for col in range(cols):
            patch = pixel_grid[row:row + 3, col:col + 3]
            activation[row, col] = np.sum(patch * kernel)

    return np.abs(activation)

# Compute two activation maps.
vertical_map = convolve_valid(image, vertical_filter)
horizontal_map = convolve_valid(image, horizontal_filter)

# Summarize the strongest responses briefly.
print("Image grid shape:", image.shape)
print("Activation map shape:", vertical_map.shape)
print("Vertical edge maximum:", round(float(vertical_map.max()), 2))
print("Horizontal edge maximum:", round(float(horizontal_map.max()), 2))

# Plot original pixels and activation maps.
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
plots = [(image, "Synthetic pavement pixels"), (vertical_map, "Vertical edge activation"), (horizontal_map, "Horizontal edge activation")]

# Display each grid with consistent styling.
for axis, item in zip(axes, plots):
    grid, title = item
    axis.imshow(grid, cmap="gray", interpolation="nearest")
    axis.set_title(title)

    axis.set_xticks([])
    axis.set_yticks([])

# Finish the single teaching figure.
plt.tight_layout()
plt.show()

```

### **3.3. Filter Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_08/Lecture_A/image_03_03.jpg?v=1783625666" width="250">



>* Visualize filters to reveal image changes
>* Compare outputs to understand useful patterns

>* Filters create maps of visual responses
>* Brightness shows pattern match strength

>* Adjust display settings to reveal filter patterns
>* Interpret filtered images with engineering judgment



```
#@title Python Code - Filter Visualization

# Visualize filters on a tiny pavement image.
# Manual convolution reveals local pixel patterns.
# One plot compares several filter responses.

import numpy as np
import matplotlib.pyplot as plt

# Create a small synthetic pavement intensity grid.
image = np.full((18, 18), 0.55, dtype=float)
image[8:10, 2:16] = 0.25
image[:, 12:14] = 0.75

# Add a bright lane marking region.
image[3:15, 4:6] = 0.90
image[13:16, 13:16] = 0.35
assert image.shape == (18, 18)

# Define simple civil image filters.
vertical_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
horizontal_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
blur_filter = np.ones((3, 3)) / 9.0

# Apply one valid convolution manually.
def apply_filter(grid, kernel):
    rows, cols = grid.shape
    krows, kcols = kernel.shape
    assert krows == 3 and kcols == 3

    output = np.zeros((rows - 2, cols - 2), dtype=float)
    for row in range(rows - 2):
        for col in range(cols - 2):
            patch = grid[row:row + 3, col:col + 3]
            output[row, col] = np.sum(patch * kernel)

    return output

# Compute response maps for each filter.
vertical_response = apply_filter(image, vertical_filter)
horizontal_response = apply_filter(image, horizontal_filter)
blur_response = apply_filter(image, blur_filter)

# Print a short interpretation summary.
print("Manual filters convert pixels into response maps.")
print("Bright edge responses mark strong local changes.")
print("Blur keeps broad regions while softening details.")

# Display original and filtered results together.
fig, axes = plt.subplots(1, 4, figsize=(12, 3))
titles = ["Original", "Vertical edges", "Horizontal edges", "Blur"]
images = [image, vertical_response, horizontal_response, blur_response]

# Use consistent grayscale visualization for comparison.
for axis, title, shown in zip(axes, titles, images):
    axis.imshow(shown, cmap="gray")
    axis.set_title(title)
    axis.axis("off")

plt.tight_layout()
plt.show()

```
