### **1.1. Function Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1783509576" width="250">



>* Functions name reusable engineering calculations
>* Clear functions make scripts easier to review

>* Give each function one clear job
>* Separate steps to simplify debugging

>* Name functions clearly for repeatable engineering tasks
>* Build trusted steps for larger analyses



```
#@title Python Code - Function Basics

# Functions make engineering calculations reusable.
# Small datasets help beginners verify results.
# Clear names explain each calculation.

# Define concrete cylinder strengths in megapascals.
strengths_mpa = [31.2, 34.8, 29.9, 36.1]

# Define pipe segment lengths in feet.
pipe_lengths_ft = [12.5, 18.0, 9.5, 14.0]

# Return the average of numeric values.
def average_value(values):
    if len(values) == 0:
        return None

    total_value = sum(values)
    average_result = total_value / len(values)
    return average_result

# Return the total of numeric values.
def total_value(values):
    if len(values) == 0:
        return 0

    total_result = sum(values)
    return total_result

# Use functions on small engineering datasets.
avg_strength = average_value(strengths_mpa)
total_pipe = total_value(pipe_lengths_ft)

# Format results for readable reporting.
strength_text = f"Average strength: {avg_strength:.1f} MPa"
pipe_text = f"Total pipe length: {total_pipe:.1f} ft"

# Print concise results for quick checking.
print(strength_text)
print(pipe_text)
print("Function names describe the engineering task.")

```

### **1.2. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1783509580" width="250">



>* Choose clear inputs for each function
>* Explicit inputs improve testing and reuse

>* Match inputs to the function’s task
>* Keep functions focused and readable

>* Use meaningful names for function inputs
>* Keep input forms clear and predictable



```
#@title Python Code - Function Inputs

# Function inputs make engineering scripts readable.
# Explicit arguments reduce hidden data dependencies.
# Small datasets keep examples easy.

# Define a focused function with clear inputs.
def summarize_core_thickness(thickness_inches, minimum_inches):
    if not thickness_inches:
        return {"average": None, "thin_count": 0}

    # Compute values using only provided inputs.
    average_value = sum(thickness_inches) / len(thickness_inches)
    thin_values = [value for value in thickness_inches if value < minimum_inches]

    # Return a small dictionary for reporting.
    return {"average": average_value, "thin_count": len(thin_values)}

# Store a tiny pavement core dataset.
core_thicknesses = [7.8, 8.1, 7.4, 8.0, 7.6]
required_thickness = 7.75

# Call the function with explicit inputs.
summary = summarize_core_thickness(core_thicknesses, required_thickness)

# Print concise results for review.
print("Pavement core thickness summary")
print("Input measurements:", core_thicknesses)
print("Required thickness:", required_thickness, "inches")
print("Average thickness:", round(summary["average"], 2), "inches")
print("Cores below requirement:", summary["thin_count"])

# Reuse the same function for another project.
second_project = [18.9, 19.4, 18.7, 19.1]
metric_limit = 19.0

# The input names show the engineering meaning.
metric_summary = summarize_core_thickness(second_project, metric_limit)
print("Second project average:", round(metric_summary["average"], 2), "cm")

```

### **1.3. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1783509578" width="250">



>* Choose inputs that match the engineering task
>* Clear inputs improve testing and reuse

>* Choose inputs that fit the task
>* Keep each function focused and reusable

>* Use clear names for expected inputs
>* Plan for missing or unusual data



```
#@title Python Code - Function Inputs

# Functions need clear input values.
# Civil datasets often stay small.
# Readable names reduce debugging time.

# Store concrete strengths in megapascals.
strengths_mpa = [31.2, 34.8, 29.5, 36.1]

# Store pavement cores in inches.
core_thickness_inches = [5.8, 6.1, 5.4, 6.0]

# Define a function with explicit inputs.
def summarize_measurements(values, unit_label, minimum_allowed):
    # Check that measurements were provided.
    if len(values) == 0:
        return {"count": 0, "average": None, "flagged": []}

    # Keep only realistic nonnegative values.
    clean_values = [value for value in values if value >= 0]

    # Avoid division when all values fail.
    if len(clean_values) == 0:
        return {"count": 0, "average": None, "flagged": []}

    # Calculate a simple engineering summary.
    average_value = sum(clean_values) / len(clean_values)

    # Flag values below the project criterion.
    flagged_values = [value for value in clean_values if value < minimum_allowed]

    # Return named results for later use.
    return {"count": len(clean_values), "average": average_value, "flagged": flagged_values}

# Call the function with concrete inputs.
concrete_summary = summarize_measurements(strengths_mpa, "MPa", 30.0)

# Call the same function with pavement inputs.
pavement_summary = summarize_measurements(core_thickness_inches, "inches", 5.5)

# Print compact results for comparison.
print("Concrete average MPa:", round(concrete_summary["average"], 2))

# Print flagged concrete measurements only.
print("Concrete flagged values:", concrete_summary["flagged"])

# Print compact pavement results too.
print("Pavement average inches:", round(pavement_summary["average"], 2))

# Print flagged pavement measurements only.
print("Pavement flagged values:", pavement_summary["flagged"])

# Show why explicit inputs help reuse.
print("Same function handled two datasets safely.")

```

### **2.1. Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1783509587" width="250">



>* Store related engineering values in order
>* Process datasets as one coherent object

>* Modify lists while cleaning engineering data
>* Use positions carefully with consistent contents

>* Apply repeated operations with loops and comprehensions
>* Build clearer, reliable engineering data workflows



```
#@title Python Code - Lists

# Lists keep engineering measurements ordered.
# Lists can grow during processing.
# Loops summarize repeated field observations.

# Store ordered pavement rut depths in millimeters.
rut_depths_mm = [4, 7, 12, 3, 15, 9]

# Add one new field observation safely.
new_depth_mm = 11
rut_depths_mm.append(new_depth_mm)

# Replace a corrected transcription error.
rut_depths_mm[2] = 10

# Convert every depth using a comprehension.
rut_depths_cm = [depth / 10 for depth in rut_depths_mm]

# Filter depths needing maintenance review.
review_limit_mm = 10
review_depths_mm = [depth for depth in rut_depths_mm if depth >= review_limit_mm]

# Summarize the ordered list with built-ins.
count_readings = len(rut_depths_mm)
average_depth = sum(rut_depths_mm) / count_readings

# Show compact results for learners.
print("Rut depths in millimeters:", rut_depths_mm)
print("Rut depths in centimeters:", rut_depths_cm)
print("Readings needing review:", review_depths_mm)
print("Number of readings:", count_readings)
print("Average rut depth, mm:", round(average_depth, 1))

```

### **2.2. Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1783509589" width="250">



>* Store engineering data with named keys
>* Make records clearer than position-based lists

>* Link engineering measurements to meaningful labels
>* Loop through records for readable checks

>* Use consistent keys for flexible records
>* Build searchable summaries with dictionary comprehensions



```
#@title Python Code - Dictionaries

# Dictionaries label engineering values clearly.
# Keys make records easier to read.
# Loops can summarize named fields.

# Create small civil engineering records.
culvert_records = [
    {"id": "C-101", "diameter_in": 36, "condition": "good"},
    {"id": "C-102", "diameter_in": 24, "condition": "poor"},

    {"id": "C-103", "diameter_in": 48, "condition": "fair"},
]

# Read one value using its key.
first_identifier = culvert_records[0]["id"]
print("First culvert identifier:", first_identifier)

# Count records with poor condition.
poor_count = 0
for record in culvert_records:
    if record["condition"] == "poor":
        poor_count = poor_count + 1

print("Poor condition count:", poor_count)

# Build a dictionary using comprehension.
diameter_lookup = {
    record["id"]: record["diameter_in"]
    for record in culvert_records
}

# Show the compact lookup dictionary.
print("Diameter lookup:", diameter_lookup)

# Add a calculated metric diameter field.
for record in culvert_records:
    record["diameter_cm"] = round(record["diameter_in"] * 2.54, 1)

# Print a short engineering summary.
for record in culvert_records:
    summary = f"{record['id']}: {record['diameter_cm']} cm, {record['condition']}"
    print(summary)

```

### **2.3. Sets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1783509591" width="250">



>* Sets store unique items, ignoring order
>* Useful for membership and duplicate checks

>* Compare engineering records across data sources
>* Find overlaps, gaps, and reporting issues

>* Sets find uniques, not order or counts
>* Combine containers to organize engineering records



```
#@title Python Code - Sets

# Sets store unique engineering identifiers.
# Membership checks support quick quality control.
# Set operations compare small record groups.

# Create repeated culvert identifiers from field logs.
scheduled_ids = ["C-101", "C-102", "C-103", "C-103"]

# Create identifiers actually inspected this week.
inspected_ids = ["C-102", "C-103", "C-104", "C-104"]

# Convert lists into sets to remove duplicates.
scheduled_set = set(scheduled_ids)
inspected_set = set(inspected_ids)

# Find identifiers appearing in both sources.
completed_ids = scheduled_set & inspected_set

# Find scheduled identifiers missing inspection records.
missed_ids = scheduled_set - inspected_set

# Find unexpected identifiers from field records.
extra_ids = inspected_set - scheduled_set

# Combine all identifiers involved in reconciliation.
all_ids = scheduled_set | inspected_set

# Sort sets before printing for readable output.
print("Unique scheduled culverts:", sorted(scheduled_set))
print("Unique inspected culverts:", sorted(inspected_set))
print("Completed inspections:", sorted(completed_ids))
print("Missed scheduled inspections:", sorted(missed_ids))

# Print unexpected and combined identifiers.
print("Unexpected field identifiers:", sorted(extra_ids))
print("All culverts mentioned:", sorted(all_ids))

# Check membership for one important asset.
target_id = "C-101"
print("Was C-101 inspected?", target_id in inspected_set)

```

## **3. Files and Debugging**

### **3.1. Reading Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1783509581" width="250">



>* Open engineering data files for Python processing
>* Make analysis repeatable and less manual

>* Use correct paths and expected file structure
>* Check data format, units, and meaning

>* Check paths and messy file contents
>* Inspect samples to build reliable scripts



```
#@title Python Code - Reading Local Files

# Read a small local engineering file.
# Check paths before parsing records.
# Summarize values after safe reading.

from pathlib import Path
import csv

# Create a tiny local CSV file.
file_path = Path("rainfall_station_sample.csv")

# Store daily rainfall depths in inches.
rows = [
    ["date", "station", "rainfall_in"],
    ["2026-07-01", "A12", "0.15"],

    ["2026-07-02", "A12", "0.00"],
    ["2026-07-03", "A12", "1.20"],
]

# Write the sample file locally.
with file_path.open("w", newline="") as file_object:
    writer = csv.writer(file_object)
    writer.writerows(rows)

# Confirm the file exists before reading.
if not file_path.exists():
    raise FileNotFoundError("Expected rainfall file was not created")

# Read rows into dictionaries.
with file_path.open("r", newline="") as file_object:
    reader = csv.DictReader(file_object)
    records = list(reader)

# Validate that records were loaded.
if len(records) == 0:
    raise ValueError("No rainfall records were found")

# Convert rainfall strings into floats.
rainfall_values = [float(row["rainfall_in"]) for row in records]

# Compute a simple engineering summary.
total_rainfall = sum(rainfall_values)
wet_days = sum(value > 0 for value in rainfall_values)

# Print a compact debugging summary.
print(f"File read: {file_path.name}")
print(f"Records loaded: {len(records)}")
print(f"First record: {records[0]}")
print(f"Total rainfall: {total_rainfall:.2f} inches")
print(f"Wet days counted: {wet_days}")

```

### **3.2. Writing Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1783509583" width="250">



>* Save script results for later use
>* Choose clear formats and meaningful filenames

>* Preserve data meaning with clear labels
>* Choose create or append carefully

>* Check folders, permissions, names, and formats
>* Verify outputs to catch hidden file errors



```
#@title Python Code - Writing Local Files

# Write small engineering results safely.
# Use clear names and headings.
# Verify files after writing.

from pathlib import Path
import csv

# Store outputs in a local folder.
output_folder = Path("civil_outputs")
output_folder.mkdir(exist_ok=True)

# Define a small inspection dataset.
bridge_records = [
    {"bridge_id": "B-101", "span_ft": 85, "rating": "Good"},
    {"bridge_id": "B-102", "span_ft": 120, "rating": "Fair"},
    {"bridge_id": "B-103", "span_ft": 60, "rating": "Poor"},
]

# Choose a meaningful output filename.
csv_path = output_folder / "bridge_inspection_summary.csv"
field_names = ["bridge_id", "span_ft", "rating"]

# Write headings and records consistently.
with csv_path.open("w", newline="") as file_object:
    writer = csv.DictWriter(file_object, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(bridge_records)

# Read the file back for debugging.
with csv_path.open("r", newline="") as file_object:
    saved_rows = list(csv.DictReader(file_object))

# Check that every record was saved.
expected_count = len(bridge_records)
actual_count = len(saved_rows)

# Create a short human-readable log.
log_path = output_folder / "file_writing_log.txt"
log_message = f"Saved {actual_count} bridge records to {csv_path.name}."

# Write the log as plain text.
with log_path.open("w") as file_object:
    file_object.write(log_message + "\n")

# Print concise verification results.
print(f"CSV file created: {csv_path}")
print(f"Log file created: {log_path}")
print(f"Expected rows: {expected_count}")
print(f"Saved rows: {actual_count}")
print(f"First saved bridge: {saved_rows[0]['bridge_id']}")

```

### **3.3. Debugging File Scripts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1783509585" width="250">



>* Check file assumptions before debugging
>* Small file errors can mislead engineering results

>* Trace file errors step by step
>* Check sample data before calculations

>* Design scripts to reveal file problems clearly
>* Build trust for engineering decisions



```
#@title Python Code - Debugging File Scripts

# Debugging file scripts starts with clear checks.
# Small civil datasets often contain messy rows.
# This example writes and inspects one file.

from pathlib import Path
import csv

# Create a tiny local rainfall file.
file_path = Path("rainfall_debug_example.csv")
rows = [
    ["station", "day", "rain_mm"],

    ["A12", "Monday", "12.5"],
    ["A12", "Tuesday", ""],
    ["A12", "Wednesday", "trace"],

    ["A12", "Thursday", "8.0"],
]

# Write the sample file safely.
with file_path.open("w", newline="") as file_object:
    writer = csv.writer(file_object)
    writer.writerows(rows)

# Prepare containers for valid and skipped records.
valid_rainfall = []
skipped_rows = []
required_columns = ["station", "day", "rain_mm"]

# Read the file with defensive checks.
try:
    with file_path.open("r", newline="") as file_object:
        reader = csv.DictReader(file_object)
        header = reader.fieldnames or []

        # Stop early if required columns are missing.
        missing = [name for name in required_columns if name not in header]
        if missing:
            raise ValueError(f"Missing columns: {missing}")

        # Inspect each row before converting values.
        for line_number, record in enumerate(reader, start=2):
            raw_value = record.get("rain_mm", "").strip()
            if raw_value == "":

                skipped_rows.append((line_number, "blank rainfall"))
                continue
            try:
                rainfall = float(raw_value)
            except ValueError:

                skipped_rows.append((line_number, f"bad value {raw_value}"))
                continue
            valid_rainfall.append(rainfall)

except FileNotFoundError:
    print(f"File not found: {file_path}")
else:
    # Summarize only after confirming valid records exist.
    if valid_rainfall:
        average_rainfall = sum(valid_rainfall) / len(valid_rainfall)
        print(f"Opened file: {file_path}")

        print(f"Valid rainfall records: {len(valid_rainfall)}")
        print(f"Skipped rows: {len(skipped_rows)}")
        print(f"Average rainfall: {average_rainfall:.1f} mm")
    else:
        print("No valid rainfall records were found.")

    # Show concise debugging details for skipped rows.
    for line_number, reason in skipped_rows:
        print(f"Skipped line {line_number}: {reason}")

print("Debugging check complete.")

```
