## **1. Readable Python Functions**

### **1.1. Function Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1783550596" width="250">



>* Functions reuse steps for repeated engineering tasks
>* Clear names make scripts easier to understand

>* Keep each function focused on one task
>* Focused functions are easier to adapt

>* Track function inputs, processing, and outputs
>* Name assumptions clearly to prevent data mistakes



```
#@title Python Code - Function Basics

# Functions make engineering calculations reusable.
# Inputs enter, results return clearly.
# Small examples support readable scripts.

# Define a focused unit conversion function.
def mm_to_inches(rainfall_mm):
    return rainfall_mm / 25.4

# Define a focused averaging function.
def average_strength(strengths_mpa):
    if len(strengths_mpa) == 0:
        return None

    return sum(strengths_mpa) / len(strengths_mpa)

# Define a focused classification function.
def classify_pavement_rating(rating):
    if rating >= 85:
        return "good"

    if rating >= 60:
        return "fair"

    return "poor"

# Store tiny civil engineering records.
site_record = {
    "site": "Bridge A",
    "rainfall_mm": 42.0,
    "strengths_mpa": [31.5, 33.0, 32.2],
    "pavement_rating": 78,
}

# Call functions using record values.
rainfall_inches = mm_to_inches(site_record["rainfall_mm"])
avg_strength = average_strength(site_record["strengths_mpa"])
pavement_class = classify_pavement_rating(site_record["pavement_rating"])

# Print a compact engineering summary.
print("Readable function example for", site_record["site"])
print("Rainfall:", round(rainfall_inches, 2), "inches")
print("Average concrete strength:", round(avg_strength, 1), "MPa")
print("Pavement condition:", pavement_class)

```

### **1.2. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1783550598" width="250">



>* Choose clear inputs for engineering functions
>* Explicit inputs improve testing and reuse

>* Name inputs by engineering meaning
>* Group many related inputs into records

>* State assumptions about data quality clearly
>* Handle messy inputs or clean them first



```
#@title Python Code - Function Inputs

# Functions need clear engineering inputs.
# Names should reveal measurement meaning.
# Small checks make scripts safer.

# Store small pavement rut records.
rut_depths_mm = [4.2, 7.8, 11.5, 3.9, 9.1]

# Store the maintenance trigger explicitly.
maintenance_limit_mm = 8.0

# Define inputs with engineering names.
def count_excessive_ruts(rut_depths_mm, limit_mm):
    # Check that records are usable.
    if not rut_depths_mm or limit_mm < 0:
        return {"count": 0, "maximum": None}

    # Keep only numeric rut measurements.
    clean_depths = [depth for depth in rut_depths_mm if isinstance(depth, (int, float))]

    # Count measurements above the limit.
    excessive_count = sum(depth > limit_mm for depth in clean_depths)

    # Return a small summary dictionary.
    return {"count": excessive_count, "maximum": max(clean_depths)}

# Call the function with explicit inputs.
rut_summary = count_excessive_ruts(rut_depths_mm, maintenance_limit_mm)

# Print a compact engineering result.
print("Rut records checked:", len(rut_depths_mm))
print("Maintenance limit, mm:", maintenance_limit_mm)
print("Ruts above limit:", rut_summary["count"])
print("Maximum rut depth, mm:", rut_summary["maximum"])
print("Clear inputs make the function reusable.")

```

### **1.3. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1783550600" width="250">



>* Choose inputs that show function purpose
>* Clear inputs improve testing and reuse

>* Name inputs by engineering meaning
>* Prevent confusion between similar data structures

>* Keep each function focused and limited
>* Make required inputs and defaults clear



```
#@title Python Code - Function Inputs

# Functions need clear engineering inputs.
# Input names should show meaning.
# Small examples make contracts visible.

# Store tiny pavement inspection records.
pavement_records = [
    {"road": "A1", "rut_depth_mm": 8, "cracks": 3},
    {"road": "A1", "rut_depth_mm": 14, "cracks": 7},

    {"road": "B2", "rut_depth_mm": 5, "cracks": 1},
    {"road": "B2", "rut_depth_mm": 11, "cracks": 4},
]

# Define inputs with engineering meaning.
def average_rut_depth(records, road_id):
    # Select records for one road.
    matching_depths = [
        record["rut_depth_mm"]
        for record in records

        if record["road"] == road_id
    ]

    # Guard against missing road identifiers.
    if len(matching_depths) == 0:
        return None

    # Return one focused summary value.
    return sum(matching_depths) / len(matching_depths)

# Define a threshold as explicit input.
def needs_review(records, road_id, limit_mm=10):
    # Reuse the focused summary function.
    average_depth = average_rut_depth(records, road_id)

    # Handle missing data safely.
    if average_depth is None:
        return "no records found"

    # Compare against the project limit.
    if average_depth > limit_mm:
        return "review needed"

    # Return a clear status otherwise.
    return "within limit"

# Call functions with visible inputs.
a1_average = average_rut_depth(pavement_records, "A1")
b2_status = needs_review(pavement_records, "B2", limit_mm=9)

# Print concise teaching results.
print("A1 average rut depth:", round(a1_average, 1), "mm")
print("B2 review status:", b2_status)
print("Missing road status:", needs_review(pavement_records, "C3"))

```

## **2. Core Python Containers**

### **2.1. Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1783550607" width="250">



>* Store related engineering values in order
>* Use positions to represent meaningful sequences

>* Lists can change as data is processed
>* Combine lists with dictionaries for clear records

>* Use loops to process list items repeatedly
>* Comprehensions create cleaner, adaptable engineering datasets



```
#@title Python Code - Lists

# Lists store ordered engineering measurements.
# Loops process each list item.
# Comprehensions create cleaned derived lists.

# Store daily rainfall depths in millimeters.
rainfall_mm = [0.0, 12.5, 8.0, 0.0, 21.3, 5.2, 0.0]

# Store matching weekday labels in order.
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Check both lists align correctly.
if len(rainfall_mm) != len(days):
    raise ValueError("Rainfall and day lists must match.")

# Build paired records using a loop.
wet_day_records = []
for index, depth in enumerate(rainfall_mm):
    if depth > 0.0:
        wet_day_records.append((days[index], depth))

# Convert millimeters to inches using comprehension.
rainfall_in = [round(depth / 25.4, 2) for depth in rainfall_mm]

# Filter heavy rainfall days using comprehension.
heavy_days = [days[i] for i, depth in enumerate(rainfall_mm) if depth >= 10.0]

# Summarize the ordered list results.
total_mm = round(sum(rainfall_mm), 1)
max_depth = max(rainfall_mm)
max_day = days[rainfall_mm.index(max_depth)]

# Print compact teaching output.
print("Rainfall list:", rainfall_mm)
print("Wet day records:", wet_day_records)
print("Rainfall in inches:", rainfall_in)
print("Heavy rainfall days:", heavy_days)
print("Weekly total mm:", total_mm)
print("Maximum day and depth:", max_day, max_depth)

```

### **2.2. Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1783550611" width="250">



>* Store engineering data as named key-value pairs
>* Access records clearly without relying on order

>* Store repeated records as named fields
>* Keep engineering details connected for summaries

>* Use clear, consistent dictionary key names
>* Loop and comprehend records for summaries



```
#@title Python Code - Dictionaries

# Dictionaries store named engineering record fields.
# Lists can hold repeated dictionary records.
# Loops summarize values by meaningful keys.

# Create small retaining wall inspection records.
wall_records = [
    {"wall_id": "RW-01", "height_ft": 12.5, "rating": 7},
    {"wall_id": "RW-02", "height_ft": 18.0, "rating": 5},
    {"wall_id": "RW-03", "height_ft": 9.5, "rating": 8},
]

# Access one value using its descriptive key.
first_height = wall_records[0]["height_ft"]
print("First wall height:", first_height, "ft")

# Update a dictionary with a new calculated field.
for record in wall_records:
    record["height_m"] = round(record["height_ft"] * 0.3048, 2)

# Select records needing closer engineering review.
review_walls = [
    record["wall_id"]
    for record in wall_records
    if record["rating"] <= 5
]

# Build a compact summary dictionary.
summary = {
    "wall_count": len(wall_records),
    "average_height_ft": round(sum(r["height_ft"] for r in wall_records) / len(wall_records), 1),
    "review_needed": review_walls,
}

# Print concise results for inspection.
print("Updated first record:", wall_records[0])
print("Walls needing review:", summary["review_needed"])
print("Average height:", summary["average_height_ft"], "ft")
print("Summary keys:", list(summary.keys()))

```

### **2.3. Sets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1783550610" width="250">



>* Sets store unique items, ignoring order
>* Useful for finding distinct engineering records

>* Compare record groups using set operations
>* Find missing items and dataset mismatches

>* Use sets when order does not matter
>* Combine containers to track unique record values



```
#@title Python Code - Sets

# Sets keep unique engineering identifiers.
# Membership checks support quick quality control.
# Set comparisons reveal missing inspection records.

# Store scheduled culvert identifiers with duplicates.
scheduled_culverts = [
    "C-101", "C-102", "C-103", "C-101",
    "C-104", "C-105", "C-106",
]

# Store completed culvert identifiers with duplicates.
completed_culverts = [
    "C-101", "C-103", "C-103", "C-107",
    "C-105",
]

# Convert lists into sets of unique identifiers.
scheduled_set = set(scheduled_culverts)
completed_set = set(completed_culverts)

# Compare sets to find useful project checks.
inspected_as_planned = scheduled_set & completed_set
still_needed = scheduled_set - completed_set
unexpected_completed = completed_set - scheduled_set

# Build small inspection records using dictionaries.
inspection_records = [
    {"culvert": "C-101", "rating": "good", "span_ft": 6},
    {"culvert": "C-103", "rating": "fair", "span_ft": 4},
    {"culvert": "C-105", "rating": "good", "span_ft": 5},
]

# Use a set comprehension for unique ratings.
unique_ratings = {
    record["rating"] for record in inspection_records
}

# Print concise results for the learner.
print("Unique scheduled culverts:", sorted(scheduled_set))
print("Inspected as planned:", sorted(inspected_as_planned))
print("Still needing inspection:", sorted(still_needed))
print("Unexpected completed culverts:", sorted(unexpected_completed))
print("Unique condition ratings:", sorted(unique_ratings))

```

## **3. Files and Debugging**

### **3.1. Reading Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1783550602" width="250">



>* Local files hold real engineering project data
>* Correct file paths prevent reading errors

>* Recognize file structure before processing data
>* Convert text into usable engineering values

>* Plan for missing or messy files
>* Check data before large processing runs



```
#@title Python Code - Reading Local Files

# Read local engineering files safely.
# Check paths before opening files.
# Convert text measurements into numbers.

from pathlib import Path
import csv

# Create a small local project file.
file_path = Path("rainfall_station_sample.csv")

# Store rainfall records as text rows.
sample_rows = [
    ["time", "station", "rainfall_in"],
    ["08:00", "A12", "0.10"],
    ["09:00", "A12", "0.25"],

    ["10:00", "A12", ""],
    ["11:00", "A12", "0.05"],
]

# Write the sample file locally.
with file_path.open("w", newline="") as file_object:
    writer = csv.writer(file_object)
    writer.writerows(sample_rows)

# Confirm the file exists first.
if not file_path.exists():
    raise FileNotFoundError("Rainfall file was not found.")

# Read rows using a dictionary reader.
valid_depths = []
skipped_rows = 0

# Open the local file safely.
with file_path.open("r", newline="") as file_object:
    reader = csv.DictReader(file_object)
    expected_columns = {"time", "station", "rainfall_in"}

    # Validate the expected column names.
    if set(reader.fieldnames or []) != expected_columns:
        raise ValueError("Unexpected rainfall file columns.")

    # Convert usable rainfall values.
    for row in reader:
        text_value = row["rainfall_in"].strip()

        # Skip blank rainfall entries.
        if text_value == "":
            skipped_rows += 1
            continue

        # Convert rainfall text into float.
        valid_depths.append(float(text_value))

# Summarize the cleaned local data.
total_rainfall = sum(valid_depths)
average_rainfall = total_rainfall / len(valid_depths)

# Print a compact debugging summary.
print(f"File read: {file_path.name}")
print(f"Valid rainfall records: {len(valid_depths)}")
print(f"Skipped incomplete rows: {skipped_rows}")
print(f"Total rainfall: {total_rainfall:.2f} inches")
print(f"Average rainfall: {average_rainfall:.2f} inches")

```

### **3.2. Writing Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1783550604" width="250">



>* Save results as reusable engineering records
>* Make outputs clear, organized, and shareable

>* Choose clear names, locations, and write modes
>* Use consistent formats, headers, units, and values

>* Verify files and sample records after writing
>* Trace outputs for transparent, repeatable results



```
#@title Python Code - Writing Local Files

# Write small engineering results safely.
# Verify files after saving them.
# Keep outputs readable and traceable.

from pathlib import Path
import csv

# Store tiny inspection records as dictionaries.
records = [
    {"bridge": "B-101", "rating": 7, "span_ft": 85},
    {"bridge": "B-102", "rating": 4, "span_ft": 120},
    {"bridge": "B-103", "rating": 6, "span_ft": 95},
]

# Choose a clear local output path.
output_path = Path("bridge_rating_summary.csv")

# Define consistent column names for sharing.
fieldnames = ["bridge", "rating", "span_ft", "needs_review"]

# Write rows using a context manager.
with output_path.open("w", newline="") as file_object:
    writer = csv.DictWriter(file_object, fieldnames=fieldnames)
    writer.writeheader()

    for record in records:
        row = record.copy()
        row["needs_review"] = record["rating"] < 5
        writer.writerow(row)

# Read back a few lines for debugging.
with output_path.open("r", newline="") as file_object:
    saved_lines = file_object.readlines()

# Check expected file size and content.
expected_lines = len(records) + 1
line_count_ok = len(saved_lines) == expected_lines
file_exists_ok = output_path.exists()

# Print a compact verification summary.
print(f"Created file: {output_path.name}")
print(f"File exists: {file_exists_ok}")
print(f"Expected lines found: {line_count_ok}")
print(f"Header: {saved_lines[0].strip()}")
print(f"First data row: {saved_lines[1].strip()}")

```

### **3.3. Debugging File Scripts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1783550605" width="250">



>* Check file assumptions before debugging
>* Small data errors can mislead analyses

>* Check each file-processing stage separately
>* Compare outputs with expected records

>* Design scripts for easier problem tracing
>* Test small examples and debug methodically



```
#@title Python Code - Debugging File Scripts

# Debugging file scripts starts with small checks.
# This example uses a tiny bridge file.
# Each stage prints one useful diagnostic.

from pathlib import Path
import csv

# Create a small local data file.
input_path = Path("bridge_ratings_debug.csv")
output_path = Path("bridge_ratings_clean.csv")

# Include one bad rating for debugging.
rows = [
    ["county", "bridge_id", "rating"],
    ["Adams", "B-101", "7"],

    ["Adams", "B-102", "note"],
    ["Clark", "B-201", "5"],
    ["Clark", "B-202", "6"],
]

# Write the sample file safely.
with input_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Confirm the expected file exists.
if not input_path.exists():
    raise FileNotFoundError("Expected bridge file was not created")

# Read records and inspect structure.
with input_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    records = list(reader)

# Validate that records were loaded.
if len(records) == 0:
    raise ValueError("Bridge file has no data rows")

# Convert ratings while collecting problems.
clean_rows = []
problems = []

# Process each bridge record carefully.
for row_number, row in enumerate(records, start=2):
    try:
        rating = int(row["rating"])

    except ValueError:
        problems.append((row_number, row["bridge_id"], row["rating"]))
        continue

    clean_rows.append([row["county"], row["bridge_id"], rating])

# Write only validated records.
with output_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["county", "bridge_id", "rating"])

    writer.writerows(clean_rows)

# Summarize the debugging results.
print(f"Input file found: {input_path.exists()}")
print(f"Rows read: {len(records)}")
print(f"Clean rows written: {len(clean_rows)}")
print(f"Problem rows found: {len(problems)}")
print(f"First problem: {problems[0] if problems else 'none'}")
print(f"Output file exists: {output_path.exists()}")

```
