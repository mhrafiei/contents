### **1.1. Function Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1783608666" width="250">



>* Reusable code for repeated engineering tasks
>* Clearer scripts with details inside functions

>* Give each function one clear job
>* Separate tasks to reduce confusion and errors

>* Test functions with small known examples
>* Make scripts clearer and easier to maintain



```
#@title Python Code - Function Basics

# Functions make engineering calculations reusable.
# Small datasets are ideal for practice.
# Clear names help colleagues review scripts.

# Define a focused conversion function.
def millimeters_to_meters(values_mm):
    if not values_mm:
        return []

    return [value / 1000 for value in values_mm]

# Define a focused averaging function.
def average_strength(strengths_mpa):
    if not strengths_mpa:
        return None

    return sum(strengths_mpa) / len(strengths_mpa)

# Define a focused classification function.
def classify_pavement(score):
    if score >= 85:
        return "good"

    if score >= 60:
        return "fair"

    return "poor"

# Store small civil engineering records.
settlement_mm = [4, 7, 9, 12]
concrete_mpa = [31.5, 34.0, 32.8]
pavement_scores = {"A1": 91, "B2": 74, "C3": 52}

# Call functions and save returned results.
settlement_m = millimeters_to_meters(settlement_mm)
mean_strength = average_strength(concrete_mpa)
condition_labels = {
    site: classify_pavement(score)
    for site, score in pavement_scores.items()
}

# Print a compact engineering summary.
print("Settlement readings in meters:", settlement_m)
print("Average concrete strength MPa:", round(mean_strength, 2))
print("Pavement condition labels:", condition_labels)
print("Functions kept each task small and readable.")

```

### **1.2. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1783608662" width="250">



>* Inputs provide data functions need
>* Clear inputs keep functions testable

>* Name inputs by engineering meaning
>* Make expected input forms clear

>* Use only necessary function inputs
>* Simpler inputs improve debugging and trust



```
#@title Python Code - Function Inputs

# Functions need clear engineering inputs.
# Names should describe civil data.
# Small inputs simplify testing work.

# Define a function with meaningful inputs.
def summarize_deflections(deflections_mm, service_limit_mm):
    valid_values = []

    # Keep only numeric measurements.
    for value in deflections_mm:
        if isinstance(value, (int, float)):
            valid_values.append(value)

    # Guard against empty usable data.
    if len(valid_values) == 0:
        return {"count": 0, "average_mm": None, "flagged": []}

    # Compute a simple engineering summary.
    average_mm = sum(valid_values) / len(valid_values)
    flagged_values = [
        value for value in valid_values
        if value > service_limit_mm
    ]

    # Return organized results for reuse.
    return {
        "count": len(valid_values),
        "average_mm": round(average_mm, 2),
        "flagged": flagged_values,
    }

# Create a tiny beam deflection dataset.
beam_deflections_mm = [8.5, 10.2, None, 13.1, 9.7]
allowable_deflection_mm = 12.0

# Call the function with explicit inputs.
summary = summarize_deflections(
    beam_deflections_mm,
    allowable_deflection_mm,
)

# Print concise results for inspection.
print("Readable function input example")
print("Measurements used:", summary["count"])
print("Average deflection, mm:", summary["average_mm"])
print("Values above limit, mm:", summary["flagged"])
print("Limit supplied, mm:", allowable_deflection_mm)

```

```
#@title Python Code - Function Inputs

# Functions need clear engineering inputs.
# Input names should show meaning.
# Small checks make functions safer.

# Define a function with focused inputs.
def summarize_deflections(deflections_mm, service_limit_mm):
    valid_values = []

    # Keep only numeric field readings.
    for reading in deflections_mm:
        if isinstance(reading, (int, float)):
            valid_values.append(reading)

    # Stop safely when no readings remain.
    if len(valid_values) == 0:
        return {"count": 0, "average_mm": None, "flagged": None}

    # Compute a simple engineering summary.
    average_mm = sum(valid_values) / len(valid_values)
    flagged = average_mm > service_limit_mm

    # Return named results for reuse.
    return {"count": len(valid_values), "average_mm": average_mm, "flagged": flagged}

# Create a tiny beam deflection dataset.
beam_deflections_mm = [8.2, 9.1, None, 10.4, "check", 7.9]
service_limit_mm = 9.0

# Call the function with explicit inputs.
summary = summarize_deflections(beam_deflections_mm, service_limit_mm)

# Print only the key teaching results.
print("Valid readings:", summary["count"])
print("Average deflection mm:", round(summary["average_mm"], 2))
print("Exceeds service limit:", summary["flagged"])

```

## **2. Core Containers**

### **2.1. Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1783608676" width="250">



>* Lists keep related engineering data ordered
>* They organize varied records for later processing

>* Process engineering measurements one item at a time
>* Turn raw observations into useful insights

>* Use loops to process list items consistently
>* Build cleaner engineering datasets with comprehensions



```
#@title Python Code - Lists

# Lists store ordered engineering measurements.
# Loops process each value consistently.
# Comprehensions create cleaned derived lists.

# Store daily rainfall depths in millimeters.
rainfall_mm = [0, 12, 8, 25, 3, 0, 18]

# Store matching inspection station labels.
stations = ["A", "B", "C", "D", "E", "F", "G"]

# Check both lists align before pairing.
if len(rainfall_mm) != len(stations):
    raise ValueError("Station and rainfall lists differ.")

# Convert millimeters to inches using comprehension.
rainfall_in = [round(value / 25.4, 2) for value in rainfall_mm]

# Select stations exceeding a drainage review threshold.
review_stations = [stations[i] for i, value in enumerate(rainfall_mm) if value >= 15]

# Accumulate total rainfall with a loop.
total_mm = 0
for value in rainfall_mm:
    total_mm = total_mm + value

# Compute average rainfall safely.
average_mm = total_mm / len(rainfall_mm)

# Print compact engineering summaries.
print("Rainfall mm:", rainfall_mm)
print("Rainfall inches:", rainfall_in)
print("Total rainfall mm:", total_mm)
print("Average rainfall mm:", round(average_mm, 1))
print("Stations needing review:", review_stations)

```

### **2.2. Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1783608674" width="250">



>* Store engineering data as key-value pairs
>* Find records by meaningful identifiers

>* Organize multi-attribute engineering records clearly
>* Nest records for reliable lookup and analysis

>* Choose stable keys and consistent value formats
>* Loop through labeled records for clear analysis



```
#@title Python Code - Dictionaries

# Dictionaries store labeled engineering records.
# Keys identify assets or samples.
# Values hold related measured information.

# Create a small bridge inspection dictionary.
bridge_ratings = {
    "BR-101": 7,
    "BR-205": 4,

    "BR-318": 6,
    "BR-422": 3,
}

# Look up one record by identifier.
selected_bridge = "BR-205"
selected_rating = bridge_ratings[selected_bridge]

# Store richer records using nested dictionaries.
bridge_records = {
    "BR-101": {"rating": 7, "span_ft": 120},
    "BR-205": {"rating": 4, "span_ft": 85},

    "BR-318": {"rating": 6, "span_ft": 140},
    "BR-422": {"rating": 3, "span_ft": 60},
}

# Select bridges needing maintenance using comprehension.
maintenance_list = [
    bridge_id
    for bridge_id, record in bridge_records.items()

    if record["rating"] <= 4
]

# Summarize ratings with a simple loop.
total_rating = 0
for record in bridge_records.values():
    total_rating += record["rating"]

# Compute the average condition rating.
record_count = len(bridge_records)
average_rating = total_rating / record_count

# Print concise teaching results.
print("Selected bridge:", selected_bridge, "rating", selected_rating)
print("Maintenance candidates:", maintenance_list)
print("Average rating:", round(average_rating, 2))

```

### **2.3. Sets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1783608679" width="250">



>* Keep only unique engineering data values
>* Use for membership, not ordered records

>* Check identifiers against expected engineering records
>* Compare datasets for overlaps and differences

>* Use sets for unique simple values
>* Compare records to spot data issues



```
#@title Python Code - Sets

# Sets keep unique engineering identifiers.
# Membership checks support data validation.
# Set comparisons reveal dataset differences.

# Create repeated bridge inspection identifiers.
bridge_ids = ["B-101", "B-102", "B-101", "B-103"]

# Convert the list into unique identifiers.
unique_bridges = set(bridge_ids)

# Define official bridge inventory identifiers.
official_bridges = {"B-101", "B-102", "B-103", "B-104"}

# Compare observed and official identifiers.
missing_bridges = official_bridges - unique_bridges
unexpected_bridges = unique_bridges - official_bridges

# Find identifiers appearing in both groups.
matched_bridges = unique_bridges & official_bridges

# Extract unique pavement material labels.
materials = ["asphalt", "concrete", "asphalt", "gravel"]
unique_materials = {item for item in materials}

# Print concise engineering set results.
print("Inspection records:", len(bridge_ids))
print("Unique bridges:", sorted(unique_bridges))
print("Matched bridges:", sorted(matched_bridges))
print("Missing official bridges:", sorted(missing_bridges))
print("Unexpected bridges:", sorted(unexpected_bridges))
print("Unique materials:", sorted(unique_materials))

```

## **3. Files and Debugging**

### **3.1. Reading Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1783608672" width="250">



>* Locate and open local engineering data safely
>* Match file format to later processing

>* Treat file rows as engineering records
>* Check columns, units, and imported values

>* Expect common file and formatting problems
>* Verify imported data before analysis



```
#@title Python Code - Reading Local Files

# Read local engineering files safely.
# Inspect records before trusting results.
# Handle missing files and bad values.

import csv
from pathlib import Path

# Create a tiny local project file.
file_path = Path("bridge_inspection_sample.csv")

# Store rows as simple dictionaries.
records = [
    {"bridge_id": "B-101", "span_ft": "120", "rating": "7"},
    {"bridge_id": "B-102", "span_ft": "85", "rating": "5"},
    {"bridge_id": "B-103", "span_ft": "", "rating": "6"},
]

# Write the sample file locally.
with file_path.open("w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=records[0].keys())
    writer.writeheader()
    writer.writerows(records)

# Read records with basic checks.
loaded_records = []
missing_span_count = 0

# Open the file only if present.
if file_path.exists() and file_path.is_file():
    with file_path.open("r", newline="", encoding="utf-8") as file:
        reader = csv.DictReader(file)
        required = {"bridge_id", "span_ft", "rating"}

        # Confirm expected columns exist.
        if required.issubset(reader.fieldnames or []):
            for row in reader:
                span_text = row["span_ft"].strip()
                rating_text = row["rating"].strip()

                # Convert values carefully.
                if span_text == "":
                    missing_span_count += 1
                    continue

                span_ft = float(span_text)
                rating = int(rating_text)
                loaded_records.append((row["bridge_id"], span_ft, rating))
        else:
            print("Missing required columns in file.")
else:
    print("Local file was not found.")

# Summarize only verified records.
record_count = len(loaded_records)
average_span = sum(item[1] for item in loaded_records) / record_count

# Print a compact debugging summary.
print(f"File read: {file_path.name}")
print(f"Verified records: {record_count}")
print(f"Missing span values skipped: {missing_span_count}")
print(f"Average verified span: {average_span:.1f} ft")
print(f"First verified record: {loaded_records[0]}")

```

### **3.2. Writing Local Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1783608668" width="250">



>* Save results as durable engineering records
>* Use clear formats for future readers

>* Choose safe file names and locations
>* Check folders, permissions, and overwrite behavior

>* Format rows consistently with clear context
>* Inspect saved files to catch errors



```
#@title Python Code - Writing Local Files

# This script writes a small engineering CSV file.
# It checks output contents after saving.
# The example uses simple bridge inspection records.

from pathlib import Path
import csv

# Store tiny records as dictionaries.
records = [
    {"span_id": "A1", "crack_mm": 0.8, "rating": "good"},
    {"span_id": "A2", "crack_mm": 1.4, "rating": "monitor"},

    {"span_id": "A3", "crack_mm": 2.1, "rating": "repair"},
]

# Choose a clear local output path.
output_folder = Path("civil_outputs")
output_file = output_folder / "bridge_inspection_summary.csv"

# Create the folder when needed.
output_folder.mkdir(parents=True, exist_ok=True)
fieldnames = ["span_id", "crack_mm", "rating"]

# Write rows using a context manager.
with output_file.open("w", newline="") as file_object:
    writer = csv.DictWriter(file_object, fieldnames=fieldnames)
    writer.writeheader()

    writer.writerows(records)

# Read the saved file for debugging.
with output_file.open("r", newline="") as file_object:
    saved_lines = file_object.read().splitlines()

# Check the expected row count.
expected_lines = len(records) + 1
actual_lines = len(saved_lines)

# Report concise debugging information.
print(f"Saved file: {output_file}")
print(f"Expected lines: {expected_lines}")
print(f"Actual lines: {actual_lines}")
print(f"Header row: {saved_lines[0]}")
print(f"First data row: {saved_lines[1]}")
print("File check passed:", actual_lines == expected_lines)

```

### **3.3. Debugging File Scripts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1783608670" width="250">



>* Check file location, access, and format
>* Inspect records before blaming code logic

>* Trace text into clean usable data
>* Check each processing step before totals

>* Design scripts with clear, helpful error messages
>* Test outputs to build reliable debugging habits



```
#@title Python Code - Debugging File Scripts

# Debugging file scripts starts with checking assumptions.
# Small civil datasets make errors easier to inspect.
# Clear messages help engineers trust processed results.

from pathlib import Path
import csv

# Create a tiny bridge inspection file.
file_path = Path("bridge_defects_debug.csv")
rows = [
    ["component", "defect_count"],
    ["deck", "3"],

    ["girder", " 2 "],
    ["bearing", "NA"],
    ["joint"],
    ["deck", "1"],
]

# Write sample records for local debugging.
with file_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Prepare containers for valid and rejected records.
summary = {}
valid_records = 0
rejected_records = []

# Confirm the expected local file exists.
if not file_path.exists():
    raise FileNotFoundError("Expected inspection file was not found.")

# Read rows and validate each field.
with file_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    for line_number, row in enumerate(reader, start=2):

        # Check that both expected columns are present.
        component = row.get("component", "").strip()
        raw_count = row.get("defect_count", "")

        # Reject rows with missing fields.
        if not component or raw_count is None:
            rejected_records.append((line_number, "missing field"))
            continue

        # Convert text safely into an integer.
        try:
            defect_count = int(raw_count.strip())
        except ValueError:
            rejected_records.append((line_number, "unreadable count"))
            continue

        # Store valid counts by bridge component.
        summary[component] = summary.get(component, 0) + defect_count
        valid_records += 1

# Print a compact debugging report.
print(f"File checked: {file_path.name}")
print(f"Valid records processed: {valid_records}")
print(f"Rejected records found: {len(rejected_records)}")
print(f"Defect summary: {summary}")
print(f"Rejected details: {rejected_records}")

```
