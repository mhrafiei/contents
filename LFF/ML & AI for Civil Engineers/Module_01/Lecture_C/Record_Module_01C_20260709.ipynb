## **1. Civil Data Modalities**

### **1.1. Tabular Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1783610573" width="250">



>* Rows store observations; columns store properties
>* Tables support comparison, filtering, and modeling

>* Civil records become searchable structured tables
>* Rows describe assets, samples, or locations

>* Tables connect measurements to engineering outcomes
>* Rows and columns may hide complexity



```
#@title Python Code - Tabular Data

# Civil tables organize engineering observations.
# Rows represent samples or assets.
# Columns represent measured engineering properties.

import pandas as pd

# Create a tiny pavement condition table.
records = [
    {"segment_id": "A1", "length_m": 120, "age_years": 8, "traffic_vpd": 9500, "distress_rating": 2, "maintenance_need": "low"},
    {"segment_id": "B2", "length_m": 85, "age_years": 18, "traffic_vpd": 14200, "distress_rating": 4, "maintenance_need": "high"},
    {"segment_id": "C3", "length_m": 200, "age_years": 12, "traffic_vpd": 7600, "distress_rating": 3, "maintenance_need": "medium"},
]

# Convert records into a tabular dataframe.
pavement = pd.DataFrame(records)

# Separate identifiers, features, and target.
identifier_column = "segment_id"
feature_columns = ["length_m", "age_years", "traffic_vpd", "distress_rating"]
target_column = "maintenance_need"

# Validate the table before summarizing.
expected_columns = [identifier_column] + feature_columns + [target_column]
assert list(pavement.columns) == expected_columns
assert pavement.shape == (3, 6)

# Print a compact view of rows and columns.
print("Tabular data example: pavement segments")
print("Rows are road segments:", pavement.shape[0])
print("Columns are recorded properties:", pavement.shape[1])
print("Feature columns:", feature_columns)
print("Target column:", target_column)
print("First two records:")
print(pavement.head(2).to_string(index=False))

```

### **1.2. Time Series Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1783610571" width="250">



>* Measurements tracked in time order
>* Supports monitoring, forecasting, and anomaly detection

>* Time order reveals physical patterns
>* Sampling choices affect what engineers detect

>* Many sources create civil time series
>* Check trends, faults, and timing issues



```
#@title Python Code - Time Series Data

# Time series data preserves engineering measurement order.
# Timestamps help reveal trends and sudden changes.
# Small sensor examples support civil data thinking.

# Import common data and plotting libraries.
import pandas as pd
import matplotlib.pyplot as plt

# Create hourly timestamps for one monitoring day.
times = pd.date_range(
    start="2026-01-01 00:00",
    periods=12,
    freq="2h"
)

# Store bridge strain readings in microstrain units.
strain = [18, 19, 21, 35, 42, 31, 24, 22, 28, 39, 33, 25]

# Build a compact time series table.
data = pd.DataFrame(
    {"timestamp": times, "strain_microstrain": strain}
)

# Validate the expected row count safely.
expected_rows = 12
assert len(data) == expected_rows

# Compare ordered and shuffled interpretations.
shuffled = data.sample(
    frac=1,
    random_state=7
).reset_index(drop=True)

# Print only a few teaching summary lines.
print("Rows are samples collected over time.")
print("Columns are timestamp and bridge strain feature.")
print("First ordered timestamp:", data.loc[0, "timestamp"])
print("First shuffled timestamp:", shuffled.loc[0, "timestamp"])
print("Maximum strain:", data["strain_microstrain"].max(), "microstrain")

# Plot the ordered sequence to show temporal structure.
plt.figure(figsize=(7, 4))
plt.plot(
    data["timestamp"],
    data["strain_microstrain"],
    marker="o"
)

# Label the civil engineering time series plot.
plt.title("Bridge Strain Gauge Time Series")
plt.xlabel("Timestamp")
plt.ylabel("Strain, microstrain")
plt.xticks(rotation=35)

# Display one compact plot for learners.
plt.tight_layout()
plt.show()

```

### **1.3. Image Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1783610575" width="250">



>* Images store pixel measurements across space
>* Spatial patterns reveal defects and surface conditions

>* Images reveal visible infrastructure conditions
>* Models analyze spatial features for decisions

>* Images include thermal, spectral, depth data
>* Capture method and scale shape decisions



```
#@title Python Code - Image Data

# Image data stores spatial engineering measurements.
# Pixels can represent visible surface conditions.
# Small arrays help explain image structure.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny grayscale pavement image.
image_size = (12, 12)
pavement = np.full(image_size, 180, dtype=int)

# Add a dark diagonal crack pattern.
for row_index in range(2, 10):
    pavement[row_index, row_index] = 35
    pavement[row_index, row_index + 1] = 55

# Add a brighter repaired patch area.
pavement[7:10, 2:5] = 225

# Validate the image-like data shape.
assert pavement.shape == image_size
assert pavement.ndim == 2

# Summarize pixels without printing arrays.
print("Image type: grayscale pavement inspection patch")
print(f"Rows x columns: {pavement.shape[0]} x {pavement.shape[1]} pixels")
print(f"Pixel intensity range: {pavement.min()} to {pavement.max()}")
print("Dark pixels can indicate cracks or shadows")
print("Bright pixels can indicate patches or markings")

# Display exactly one image-style plot.
plt.figure(figsize=(5, 4))
plt.imshow(pavement, cmap="gray", vmin=0, vmax=255)

# Add labels that connect pixels to space.
plt.title("Tiny Image Data Example: Pavement Surface")
plt.xlabel("Pixel column, across pavement")
plt.ylabel("Pixel row, along pavement")

# Add a colorbar for intensity meaning.
colorbar = plt.colorbar()
colorbar.set_label("Brightness intensity")

# Show the spatial pixel arrangement.
plt.tight_layout()
plt.show()

```

## **2. Features and Targets**

### **2.1. Rows as Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1783610585" width="250">



>* Each row represents one measured sample
>* Rows define the repeated observation unit

>* Define samples based on the engineering question
>* Check row independence and decision fit

>* Many rows may not be independent evidence
>* Describe each row to clarify meaning



```
#@title Python Code - Rows as Samples

# Rows represent measured engineering samples.
# Columns store features and targets.
# Small tables make structure visible.

import pandas as pd

# Create one row per concrete cylinder.
records = [
    {"sample_id": "C1", "age_days": 7, "water_cement": 0.50, "strength_mpa": 24.1},
    {"sample_id": "C2", "age_days": 14, "water_cement": 0.45, "strength_mpa": 31.8},
    {"sample_id": "C3", "age_days": 28, "water_cement": 0.40, "strength_mpa": 39.6},
]

# Convert records into a table.
df = pd.DataFrame(records)

# Separate descriptive features from target.
features = df[["age_days", "water_cement"]]
target = df["strength_mpa"]

# Check rows match sample count.
row_count = len(df.index)
assert row_count == len(target.index)

# Print a compact teaching summary.
print("Each row is one tested concrete cylinder.")
print(f"Rows, or samples: {row_count}")
print(f"Feature columns: {list(features.columns)}")
print(f"Target column: {target.name}")
print("First sample as plain language:")

# Describe one row without printing everything.
first = df.iloc[0]
message = (
    f"{first['sample_id']}: age {first['age_days']} days, "
    f"w/c {first['water_cement']}, strength {first['strength_mpa']} MPa"
)
print(message)

```

### **2.2. Columns as Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1783610589" width="250">



>* Columns describe each engineering sample
>* Features can be measurements or categories

>* Feature definitions shape model interpretation
>* Include units, methods, and engineering meaning

>* Create features from meaningful engineering measurements
>* Choose columns carefully to avoid misleading analysis



```
#@title Python Code - Columns as Features

# Columns can represent engineering feature measurements.
# Rows represent comparable civil engineering samples.
# Targets are outcomes we may explain.

# Import pandas for small tabular examples.
import pandas as pd

# Create a tiny bridge inspection table.
bridge_data = {
    "bridge_id": ["B1", "B2", "B3", "B4"],
    "age_years": [12, 45, 28, 63],

    "span_length_ft": [80, 140, 95, 210],
    "deck_rating": [8, 5, 7, 4],
    "traffic_vehicles_day": [12000, 34000, 18000, 52000],

    "material_type": ["steel", "concrete", "steel", "concrete"],
    "needs_repair": ["no", "yes", "no", "yes"],
}

# Convert the dictionary into a table.
bridges = pd.DataFrame(bridge_data)

# Separate descriptive columns from the outcome column.
feature_columns = [
    "age_years",
    "span_length_ft",
    "deck_rating",

    "traffic_vehicles_day",
    "material_type",
]

# Select features and target using column names.
features = bridges[feature_columns]
target = bridges["needs_repair"]

# Validate the table has matching sample counts.
assert len(features) == len(target)
assert features.shape[0] == bridges.shape[0]

# Print a compact teaching summary.
print("Samples are rows:", bridges.shape[0])
print("Feature columns:", list(features.columns))
print("Target column:", target.name)
print("One sample features:", features.iloc[0].to_dict())
print("One sample target:", target.iloc[0])

# Show why units and definitions matter.
print("Span feature uses feet, not meters.")
print("Deck rating is condition information.")
print("Material type is categorical information.")
print("Each feature column describes every sample consistently.")

```

### **2.3. Targets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1783610587" width="250">



>* Targets are outcomes models aim to predict
>* Clear targets define the dataset’s purpose

>* Targets may be numbers, categories, or events
>* Choose targets that match engineering decisions

>* Align targets with rows, features, and timing
>* Account for uncertainty in engineering measurements



```
#@title Python Code - Targets

# Targets are outcomes engineers want to estimate.
# Features are measurements available before decisions.
# Rows connect one sample to one target.

# Import pandas for small tabular examples.
import pandas as pd

# Create a tiny concrete testing dataset.
data = {
    "cement_kg_m3": [320, 350, 380, 410],
    "water_kg_m3": [180, 175, 170, 165],
    "curing_days": [7, 14, 28, 28],

    "temperature_c": [20, 22, 21, 23],
    "strength_mpa": [24.5, 32.0, 41.8, 48.2],
}

# Convert the dictionary into rows and columns.
df = pd.DataFrame(data)

# Separate supporting measurements from the target.
feature_columns = [
    "cement_kg_m3", "water_kg_m3",
    "curing_days", "temperature_c",
]

target_column = "strength_mpa"
features = df[feature_columns]
target = df[target_column]

# Validate that every sample has one target.
assert len(features) == len(target)
assert target.notna().all()

# Print a compact teaching summary.
print("Samples, features:", features.shape)
print("Target column:", target_column)
print("First sample features:", features.iloc[0].to_dict())
print("First sample target:", target.iloc[0], "MPa")
print("Engineering question: predict strength from earlier measurements.")

```

## **3. Data Quality Risks**

### **3.1. Missing Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1783610577" width="250">



>* Missing values can distort engineering model patterns
>* Gaps may reflect critical conditions studied

>* Ask why data are missing
>* Systematic gaps can bias model decisions

>* Avoid automatic deletion or average filling
>* Investigate missingness patterns and document assumptions



```
#@title Python Code - Missing Data

# Missing values can bias engineering conclusions.
# Small tables reveal data quality risks.
# We inspect gaps before filling them.

import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
data = {
    "site": ["A", "B", "C", "D", "E", "F"],
    "rainfall_mm": [12, 18, None, 55, None, 70],
    "runoff_cms": [1.1, 1.8, 3.9, None, 6.2, None],
    "sensor_status": ["ok", "ok", "storm", "storm", "storm", "storm"],
}

# Convert the dictionary into a table.
df = pd.DataFrame(data)

# Count missing values by column.
missing_counts = df.isna().sum()

# Compare missingness during storm conditions.
storm_rows = df[df["sensor_status"] == "storm"]
storm_missing = storm_rows.isna().sum().sum()

# Compare missingness during normal conditions.
ok_rows = df[df["sensor_status"] == "ok"]
ok_missing = ok_rows.isna().sum().sum()

# Show only compact teaching output.
print("Rows represent sites; columns represent measurements.")
print("Missing values by column:", missing_counts.to_dict())
print("Missing cells during storms:", int(storm_missing))
print("Missing cells during normal periods:", int(ok_missing))

# Fill rainfall with the average value.
mean_rainfall = df["rainfall_mm"].mean()
df["rainfall_filled"] = df["rainfall_mm"].fillna(mean_rainfall)

# Compare original and filled rainfall variability.
original_spread = df["rainfall_mm"].std()
filled_spread = df["rainfall_filled"].std()

# Print the key modeling warning.
print("Original rainfall spread:", round(original_spread, 1))
print("Filled rainfall spread:", round(filled_spread, 1))
print("Average filling can hide extreme storm variability.")

# Plot missingness pattern by engineering condition.
plot_data = df.set_index("site")[["rainfall_mm", "runoff_cms"]].isna()
ax = plot_data.astype(int).plot(kind="bar", figsize=(7, 3))
ax.set_title("Missing Measurements by Site")
ax.set_ylabel("Missing equals one")
ax.set_xlabel("Monitoring site")
plt.tight_layout()
plt.show()

```

### **3.2. Missing Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1783610579" width="250">



>* Missing values create uncertainty in engineering datasets
>* Unrecognized gaps can mislead model learning

>* Missingness reasons shape data risk
>* Ignoring context can hide critical failures

>* Avoid automatic deletion or average filling
>* Investigate missingness as possible engineering signal



```
#@title Python Code - Missing Data

# Missing values can hide engineering risk.
# Small tables make data quality visible.
# Simple summaries support careful judgment.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny bridge monitoring table.
data = {
    "hour": [1, 2, 3, 4, 5, 6],
    "strain_microstrain": [120, 125, np.nan, np.nan, 180, 190],
    "temperature_F": [68, 69, 70, np.nan, 75, 76],
    "storm_flag": [False, False, True, True, True, False],
}

# Convert the dictionary into rows and columns.
df = pd.DataFrame(data)

# Count missing values in each feature.
missing_counts = df.isna().sum()

# Compare missingness during storm and normal hours.
storm_missing = df.loc[df["storm_flag"], "strain_microstrain"].isna().mean()
normal_missing = df.loc[~df["storm_flag"], "strain_microstrain"].isna().mean()

# Show compact results without printing the full table.
print("Rows represent hourly bridge samples.")
print("Columns are measured features and context.")
print("Missing strain values:", int(missing_counts["strain_microstrain"]))
print("Missing temperature values:", int(missing_counts["temperature_F"]))
print("Storm strain missing rate:", round(storm_missing, 2))
print("Normal strain missing rate:", round(normal_missing, 2))
print("Lesson: missingness may follow critical conditions.")

# Plot observed strain and mark missing hours.
plt.figure(figsize=(7, 4))
plt.plot(df["hour"], df["strain_microstrain"], marker="o", label="Observed strain")
plt.scatter(df.loc[df["strain_microstrain"].isna(), "hour"], [150, 150], color="red", label="Missing strain")
plt.xlabel("Hour")
plt.ylabel("Strain, microstrain")
plt.title("Missing bridge strain readings during storm hours")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Outliers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1783610581" width="250">



>* Outliers are unusual values needing investigation
>* They may be real events or errors

>* Outliers can teach models false patterns
>* Unusual samples reduce accuracy and reliability

>* Judge outliers using engineering context
>* Document decisions to protect model credibility



```
#@title Python Code - Outliers

# Outliers can distort engineering data patterns.
# Small examples make risks easier.
# We compare normal and suspicious readings.

# Import compact data and plotting tools.
import pandas as pd
import matplotlib.pyplot as plt

# Create bridge strain readings in microstrain.
strain_values = [118, 121, 119, 122, 120, 117, 123, 119, 121, 480]
reading_ids = list(range(1, len(strain_values) + 1))

# Store readings as a small table.
data = pd.DataFrame({"reading_id": reading_ids, "strain_microstrain": strain_values})
assert len(data) == len(strain_values)

# Compute simple outlier limits using quartiles.
q1 = data["strain_microstrain"].quantile(0.25)
q3 = data["strain_microstrain"].quantile(0.75)

# Calculate the interquartile range.
iqr = q3 - q1
upper_limit = q3 + 1.5 * iqr

# Flag readings above the upper limit.
data["possible_outlier"] = data["strain_microstrain"] > upper_limit
outlier_count = int(data["possible_outlier"].sum())

# Compare averages with and without outliers.
mean_all = data["strain_microstrain"].mean()
mean_clean = data.loc[~data["possible_outlier"], "strain_microstrain"].mean()

# Print a short engineering interpretation.
print("Bridge strain readings analyzed: 10 samples")
print(f"Upper outlier limit: {upper_limit:.1f} microstrain")
print(f"Possible outliers found: {outlier_count}")
print(f"Mean with outlier: {mean_all:.1f} microstrain")
print(f"Mean without outlier: {mean_clean:.1f} microstrain")
print("Investigate before deleting: sensor fault or real extreme load?")

# Plot readings and the outlier threshold.
plt.figure(figsize=(7, 4))
plt.scatter(data["reading_id"], data["strain_microstrain"], c=data["possible_outlier"].map({False: "steelblue", True: "crimson"}), s=80)
plt.axhline(upper_limit, color="orange", linestyle="--", label="Outlier limit")
plt.xlabel("Sensor reading number")
plt.ylabel("Bridge strain, microstrain")
plt.title("One unusual reading can shift engineering summaries")
plt.legend()
plt.tight_layout()
plt.show()

```
