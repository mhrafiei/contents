## **1. Civil Data Modalities**

### **1.1. Tabular Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1783551779" width="250">



>* Rows store records; columns store properties
>* Tables support comparison, filtering, and modeling

>* Civil data often becomes structured tables
>* Records share consistent measured attributes

>* Define rows and columns in engineering terms
>* Remember tables simplify complex real systems



```
#@title Python Code - Tabular Data

# Tabular data organizes engineering records clearly.
# Rows represent samples, assets, or tests.
# Columns represent measured engineering properties.

import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny concrete testing table.
records = [
    {"specimen": "C1", "age_days": 7, "w_c_ratio": 0.50,
     "strength_psi": 3100, "slump_in": 4.0},

    {"specimen": "C2", "age_days": 14, "w_c_ratio": 0.48,
     "strength_psi": 4200, "slump_in": 3.5},

    {"specimen": "C3", "age_days": 28, "w_c_ratio": 0.45,
     "strength_psi": 5200, "slump_in": 3.0},

    {"specimen": "C4", "age_days": 28, "w_c_ratio": 0.60,
     "strength_psi": None, "slump_in": 7.5},
]

# Convert records into a tabular dataframe.
df = pd.DataFrame(records)

# Separate identifiers, features, and target.
identifier_column = "specimen"
feature_columns = ["age_days", "w_c_ratio", "slump_in"]
target_column = "strength_psi"

# Count rows, columns, and missing values.
row_count, column_count = df.shape
missing_strengths = df[target_column].isna().sum()

# Print a compact teaching summary.
print(f"Rows are concrete specimens: {row_count} samples.")
print(f"Columns are recorded properties: {column_count} columns.")
print(f"Identifier column: {identifier_column}.")
print(f"Feature columns: {feature_columns}.")
print(f"Target column: {target_column}.")
print(f"Missing target values: {missing_strengths}.")

# Show only the first few table rows.
print(df.head(3).to_string(index=False))

# Plot one feature against the target.
clean_df = df.dropna(subset=[target_column])
plt.figure(figsize=(6, 4))
plt.scatter(clean_df["age_days"], clean_df[target_column])
plt.xlabel("Curing age, days")
plt.ylabel("Compressive strength, psi")
plt.title("Tabular concrete test records")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Time Series Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1783551781" width="250">



>* Measurements ordered by time
>* Changes reveal engineering behavior

>* Time patterns reflect systems and human activity
>* Patterns reveal normal and unusual behavior

>* Timestamp spacing and reliability shape interpretation
>* Data gaps can mimic engineering patterns



```
#@title Python Code - Time Series Data

# Time series data preserves measurement order.
# Civil sensors often report repeated observations.
# Timing helps reveal trends and anomalies.

import pandas as pd
import matplotlib.pyplot as plt

# Create hourly bridge strain measurements.
times = pd.date_range(start="2026-01-01 06:00", periods=12, freq="h")

# Values mimic traffic peaks and sensor trouble.
strain_micro = [12, 18, 35, 48, 42, 28, 22, 25, 55, 60, 20, 20]

# Store timestamps and measurements together.
data = pd.DataFrame({"time": times, "strain_microstrain": strain_micro})

# Check the expected small table shape.
assert data.shape == (12, 2), "Unexpected time series size"

# Compare ordered and shuffled interpretations.
shuffled = data.sample(frac=1, random_state=7).reset_index(drop=True)

# Print only a compact teaching summary.
print("Time series rows:", len(data))
print("Columns:", list(data.columns))
print("First timestamp:", data.loc[0, "time"])
print("Last timestamp:", data.loc[11, "time"])

# Show why order matters for interpretation.
print("Ordered first three strains:", data["strain_microstrain"].head(3).tolist())
print("Shuffled first three strains:", shuffled["strain_microstrain"].head(3).tolist())

# Plot strain against time to reveal patterns.
plt.figure(figsize=(8, 4))
plt.plot(data["time"], data["strain_microstrain"], marker="o")
plt.title("Bridge Strain Sensor Time Series")
plt.xlabel("Time of day")

# Label engineering units clearly.
plt.ylabel("Strain, microstrain")
plt.grid(True, alpha=0.3)
plt.xticks(rotation=30)
plt.tight_layout()

# Display exactly one plot.
plt.show()

```

### **1.3. Image Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1783551783" width="250">



>* Images store spatial measurements as pixel grids
>* Visual patterns reveal measurable infrastructure conditions

>* Images come from many civil engineering sensors
>* Spatial location gives measurements engineering meaning

>* Images require spatial pattern interpretation
>* Quality depends on capture conditions



```
#@title Python Code - Image Data

# Image data stores measurements on spatial grids.
# Nearby pixels often describe nearby infrastructure locations.
# This example creates a tiny pavement image.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a repeatable synthetic pavement surface.
rng = np.random.default_rng(seed=7)
base = rng.normal(loc=0.55, scale=0.06, size=(40, 40))

# Add a dark diagonal crack pattern.
for row in range(8, 34):
    col = row - 3
    base[row, col:col + 2] = 0.08

# Add a bright painted lane marking.
base[5:35, 30:33] = 0.92
image = np.clip(base, 0.0, 1.0)

# Summarize image structure without printing arrays.
height, width = image.shape
pixel_count = image.size
print(f"Image grid: {height} rows by {width} columns.")

# Connect pixels to engineering interpretation.
print(f"Each sample here is one pixel measurement.")
print(f"Dark crack pixels are below brightness 0.20.")
print(f"Bright marking pixels are above brightness 0.85.")

# Count simple visual categories.
crack_pixels = int(np.sum(image < 0.20))
marking_pixels = int(np.sum(image > 0.85))
print(f"Detected dark pixels: {crack_pixels}.")
print(f"Detected bright pixels: {marking_pixels}.")

# Display exactly one image-style plot.
fig, ax = plt.subplots(figsize=(5, 5))
shown = ax.imshow(image, cmap="gray", vmin=0, vmax=1)

# Label the spatial grid clearly.
ax.set_title("Synthetic pavement image data")
ax.set_xlabel("Column position, centimeters")
ax.set_ylabel("Row position, centimeters")

# Add a compact color scale.
fig.colorbar(shown, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()

```

## **2. Features and Targets**

### **2.1. Rows as Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1783551773" width="250">



>* Each row represents one measured sample
>* Rows collect measurements for learning patterns

>* Rows define each engineering sample clearly
>* Measurements connect to one real-world item

>* Row definitions shape every analysis
>* Avoid mixing observation levels



```
#@title Python Code - Rows as Samples

# Rows represent individual engineering samples.
# Columns store measurements for each sample.
# Targets are outcomes we may predict.

import pandas as pd

# Create a tiny bridge inspection table.
bridge_data = {
    "bridge_id": ["B-101", "B-102", "B-103"],
    "age_years": [42, 18, 67],
    "span_ft": [120, 85, 210],
    "deck_rating": [6, 8, 4],
    "needs_repair": ["yes", "no", "yes"],
}

# Convert the dictionary into rows and columns.
bridges = pd.DataFrame(bridge_data)

# Separate descriptive features from the target outcome.
feature_columns = ["age_years", "span_ft", "deck_rating"]
target_column = "needs_repair"

# Count samples and measurements in the table.
sample_count = len(bridges)
feature_count = len(feature_columns)

# Select one row as one engineering sample.
one_sample = bridges.loc[1, ["bridge_id"] + feature_columns]

# Print a compact teaching summary.
print("Each row is one bridge inspection sample.")
print(f"Samples, or rows: {sample_count}")
print(f"Feature columns: {feature_count}")
print(f"Target column: {target_column}")
print("Example sample values:")
print(one_sample.to_string())

```

### **2.2. Columns as Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1783551777" width="250">



>* Columns describe each engineering sample
>* Features organize measurements for model learning

>* Feature choices shape model learning
>* Domain knowledge identifies meaningful measurements

>* Check how each feature was measured
>* Watch for inconsistent or misleading columns



```
#@title Python Code - Columns as Features

# Columns become features for engineering samples.
# Each row represents one inspected asset.
# Targets are outcomes we may explain.

import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny bridge inspection table.
data = {
    "bridge_id": ["B1", "B2", "B3", "B4"],
    "age_years": [12, 45, 28, 60],

    "span_ft": [80, 140, 95, 210],
    "traffic_vehicles_day": [9000, 22000, 15000, 31000],
    "deck_rating": [8, 5, 6, 4],
}

# Convert the dictionary into rows and columns.
bridges = pd.DataFrame(data)

# Separate descriptive columns from the outcome column.
feature_columns = ["age_years", "span_ft", "traffic_vehicles_day"]
target_column = "deck_rating"

# Select features and target using column names.
features = bridges[feature_columns]
target = bridges[target_column]

# Check that rows match across features and target.
assert len(features) == len(target)

# Print a compact teaching summary.
print("Rows are bridge inspection samples:", len(bridges))
print("Feature columns describe each sample:", feature_columns)
print("Target column is the outcome:", target_column)
print("One sample as feature values:")
print(features.iloc[0].to_dict())

# Plot one feature against the target.
plt.figure(figsize=(6, 4))
plt.scatter(bridges["age_years"], bridges["deck_rating"], s=80)
plt.xlabel("Bridge age, years")
plt.ylabel("Deck condition rating")
plt.title("One feature compared with one target")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Targets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1783551775" width="250">



>* Targets are outcomes models try to predict
>* They guide analysis toward engineering decisions

>* Define targets as meaningful engineering measurements
>* Poor targets can mislead model decisions

>* Targets can contain errors or bias
>* Avoid using future information as features



```
#@title Python Code - Targets

# Targets define the engineering outcome.
# Features describe measurements available earlier.
# Rows represent individual engineering samples.

import pandas as pd

# Create a tiny bridge inspection dataset.
data = {
    "bridge_id": ["B1", "B2", "B3", "B4"],
    "age_years": [12, 45, 30, 8],

    "span_feet": [80, 140, 95, 60],
    "avg_strain_microstrain": [120, 310, 220, 90],
    "damage_rating": ["low", "high", "medium", "low"],
}

# Convert the dictionary into rows and columns.
df = pd.DataFrame(data)

# Separate inputs from the target outcome.
feature_columns = ["age_years", "span_feet", "avg_strain_microstrain"]
target_column = "damage_rating"

# Count samples and feature measurements.
sample_count = len(df)
feature_count = len(feature_columns)

# Show the dataset structure without printing everything.
print("Samples, or rows:", sample_count)
print("Feature columns:", feature_columns)
print("Target column:", target_column)
print("First target values:", df[target_column].head(3).tolist())

# Compare one sample's features with its target.
first_features = df.loc[0, feature_columns].to_dict()
first_target = df.loc[0, target_column]

# Print a compact engineering interpretation.
print("Sample B1 features:", first_features)
print("Sample B1 target:", first_target)
print("Question learned: features predict the target.")

```

## **3. Data Quality Risks**

### **3.1. Missing Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1783551771" width="250">



>* Missing measurements create hidden data gaps
>* Incomplete data can bias model predictions

>* Missingness reasons can shape model bias
>* Gaps may hide critical conditions

>* Avoid automatic deletion or careless value filling
>* Investigate gaps to reduce bias and uncertainty



```
#@title Python Code - Missing Data

# Missing data can bias engineering conclusions.
# Small tables reveal hidden data quality risks.
# We inspect gaps before modeling decisions.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny bridge monitoring table.
data = {
    "hour": list(range(1, 9)),
    "strain_microstrain": [120, 125, np.nan, np.nan, 180, 190, 195, 200],
    "temperature_C": [18, 18.5, 19, np.nan, 24, 25, 25.5, 26],
    "traffic_count": [40, 42, 45, 80, np.nan, np.nan, 120, 130],
}

# Convert the dictionary into tabular data.
df = pd.DataFrame(data)

# Count missing values in each column.
missing_counts = df.isna().sum()
missing_rates = (df.isna().mean() * 100).round(1)

# Compare deletion with simple filling.
rows_before = len(df)
rows_after_drop = len(df.dropna())
filled_traffic_mean = df["traffic_count"].fillna(df["traffic_count"].mean())

# Print a compact missing data summary.
print("Missing data risk check for bridge monitoring")
print(f"Rows before deletion: {rows_before}")
print(f"Rows after deleting incomplete rows: {rows_after_drop}")
print("Missing counts:", missing_counts.to_dict())
print("Missing rates (%):", missing_rates.to_dict())
print(f"Mean-filled traffic count: {filled_traffic_mean.mean():.1f}")
print("Question: are gaps random, or during critical conditions?")

# Plot missingness patterns across time.
missing_map = df.set_index("hour").isna()
plt.figure(figsize=(7, 3))
plt.imshow(missing_map.T, aspect="auto", cmap="Reds")
plt.yticks(range(len(missing_map.columns)), missing_map.columns)
plt.xticks(range(len(missing_map.index)), missing_map.index)
plt.xlabel("Hour")
plt.title("Missing readings by hour and feature")
plt.colorbar(label="Missing equals one")
plt.tight_layout()
plt.show()

```

### **3.2. Missing Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1783551767" width="250">



>* Missing measurements create hidden data gaps
>* Models may learn an incomplete system picture

>* Missing data may reflect extreme conditions
>* Ignoring gaps can make models misleading

>* Avoid automatic deletion or careless filling
>* Document patterns and uncertainty transparently



```
#@title Python Code - Missing Data

# Missing values can hide important engineering conditions.
# Small tables help reveal risky data gaps.
# This example uses bridge sensor readings.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny bridge monitoring table.
data = {
    "hour": list(range(1, 9)),
    "strain_microstrain": [120, 125, np.nan, np.nan, 180, 190, 130, 128],
    "temperature_F": [68, 69, np.nan, np.nan, 82, 85, 70, 69],
    "wind_mph": [8, 9, 35, 42, 18, 16, np.nan, 7],
}

# Convert the dictionary into tabular data.
df = pd.DataFrame(data)

# Count missing values in each column.
missing_counts = df.isna().sum()
missing_percent = (missing_counts / len(df) * 100).round(1)

# Summarize missingness without printing the full table.
summary = pd.DataFrame({"missing_count": missing_counts, "missing_percent": missing_percent})
print("Missing data summary by column:")
print(summary.to_string())

# Compare averages before and after dropping missing rows.
original_mean = df["wind_mph"].mean()
clean_mean = df.dropna()["wind_mph"].mean()

# Show how deletion can change engineering interpretation.
print("\nAverage wind speed using available readings:", round(original_mean, 1), "mph")
print("Average wind speed after deleting incomplete rows:", round(clean_mean, 1), "mph")
print("Rows kept after deletion:", len(df.dropna()), "of", len(df))

# Plot missing values as a compact visual pattern.
missing_map = df.isna().astype(int)
plt.figure(figsize=(7, 3))
plt.imshow(missing_map.T, aspect="auto", cmap="Reds")
plt.yticks(range(len(df.columns)), df.columns)
plt.xticks(range(len(df)), df["hour"])
plt.xlabel("Hour")
plt.title("Missing Data Pattern in Bridge Monitoring")
plt.colorbar(label="Missing equals one")
plt.tight_layout()
plt.show()

```

### **3.3. Outliers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1783551769" width="250">



>* Outliers are unusual values needing investigation
>* They may signal events or data errors

>* Extreme values can distort learned patterns
>* Distorted models can guide poor decisions

>* Judge outliers using engineering context
>* Investigate causes before removing unusual data



```
#@title Python Code - Outliers

# Outliers can distort engineering data patterns.
# Small examples make data risks visible.
# Always investigate before deleting unusual values.

import pandas as pd
import matplotlib.pyplot as plt

# Create daily bridge strain readings.
days = list(range(1, 11))

# One reading is suspiciously high.
strain = [118, 121, 119, 122, 120, 450, 123, 121, 119, 122]

# Store measurements in a small table.
data = pd.DataFrame({"day": days, "strain_microstrain": strain})

# Check the table size safely.
assert data.shape == (10, 2)

# Compute simple summary values.
median_value = data["strain_microstrain"].median()
mean_value = data["strain_microstrain"].mean()

# Flag values far from typical behavior.
threshold = median_value * 2
outliers = data[data["strain_microstrain"] > threshold]

# Print concise teaching results.
print(f"Median strain: {median_value:.1f} microstrain")
print(f"Mean strain: {mean_value:.1f} microstrain")
print(f"Outlier threshold: {threshold:.1f} microstrain")
print(f"Flagged outlier days: {outliers['day'].tolist()}")
print("Question: sensor error, rare event, or new condition?")

# Plot readings and the outlier threshold.
plt.figure(figsize=(7, 4))

# Show the time-series measurements.
plt.plot(data["day"], data["strain_microstrain"], marker="o")

# Add a visual investigation threshold.
plt.axhline(threshold, color="red", linestyle="--", label="outlier threshold")

# Label the engineering context clearly.
plt.title("Bridge Strain Readings With One Outlier")
plt.xlabel("Day")
plt.ylabel("Strain, microstrain")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```
