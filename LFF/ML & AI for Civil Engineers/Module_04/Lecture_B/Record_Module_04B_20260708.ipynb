## **1. Feature Matrix Setup**

### **1.1. Scaled Feature Matrix**

>* Scaling balances different units and ranges

>* Select, clean, and align numeric features
>* Scale each column for regression training

```
#@title Python Code - Scaled Feature Matrix

# Build a small civil engineering dataset.
# Select several numeric regression features.
# Scale columns using training statistics.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create deterministic sample measurements.
np.random.seed(7)
rows = 12

# Store mixed-unit bridge inspection features.
data = pd.DataFrame({
    "span_ft": np.linspace(40, 180, rows),
    "deck_width_m": np.linspace(7, 18, rows),
    "age_years": np.arange(5, 65, 5),
})

# Add a simple target variable.
data["repair_cost_k"] = (
    0.8 * data["span_ft"]
    + 4.0 * data["deck_width_m"]
    + 1.5 * data["age_years"]
)

# Select feature columns and target column.
feature_cols = ["span_ft", "deck_width_m", "age_years"]
target_col = "repair_cost_k"

# Split rows without shuffling for clarity.
train_df = data.iloc[:8].copy()
test_df = data.iloc[8:].copy()

# Convert selected features into matrices.
X_train_raw = train_df[feature_cols].to_numpy(dtype=float)
X_test_raw = test_df[feature_cols].to_numpy(dtype=float)

# Validate matrix shapes before scaling.
assert X_train_raw.shape == (8, 3)
assert X_test_raw.shape == (4, 3)

# Learn scaling values from training data.
train_mean = X_train_raw.mean(axis=0)
train_std = X_train_raw.std(axis=0)

# Prevent division by zero safely.
train_std = np.where(train_std == 0, 1, train_std)
X_train_scaled = (X_train_raw - train_mean) / train_std

# Apply identical scaling to test data.
X_test_scaled = (X_test_raw - train_mean) / train_std
scaled_df = pd.DataFrame(X_train_scaled, columns=feature_cols)

# Print compact teaching outputs.
print("Raw training matrix shape:", X_train_raw.shape)
print("Scaled training matrix shape:", X_train_scaled.shape)
print("Training means after scaling:", np.round(X_train_scaled.mean(axis=0), 2))
print("Training stds after scaling:", np.round(X_train_scaled.std(axis=0), 2))
print("First scaled row:", np.round(X_train_scaled[0], 2))

# Plot raw versus scaled span values.
plt.figure(figsize=(6, 4))
plt.plot(X_train_raw[:, 0], label="raw span_ft", marker="o")
plt.plot(X_train_scaled[:, 0], label="scaled span_ft", marker="s")
plt.title("Raw and Scaled Feature Example")
plt.xlabel("Training row index")
plt.ylabel("Feature value")
plt.legend()
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Feature Scaling

# Demonstrate scaling for multiple regression features.
# Use pandas inspection and NumPy transformation.
# Keep training data statistics separate.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
data = pd.DataFrame({
    "floor_area_sqft": [850, 1200, 1500, 1800, 2200, 2600],
    "distance_miles": [1.2, 3.5, 2.1, 5.0, 4.2, 6.8],
    "building_age_years": [5, 18, 12, 30, 22, 40],
    "monthly_energy_kwh": [620, 760, 810, 980, 1050, 1240],
})

# Select input columns and target column.
feature_cols = ["floor_area_sqft", "distance_miles", "building_age_years"]
target_col = "monthly_energy_kwh"

# Split rows before learning scaling values.
train_df = data.iloc[:4].copy()
test_df = data.iloc[4:].copy()

# Convert selected columns into NumPy arrays.
X_train_raw = train_df[feature_cols].to_numpy(dtype=float)
X_test_raw = test_df[feature_cols].to_numpy(dtype=float)

# Validate feature matrices before scaling.
assert X_train_raw.shape == (4, 3)
assert X_test_raw.shape == (2, 3)

# Learn scaling values from training data only.
train_mean = X_train_raw.mean(axis=0)
train_std = X_train_raw.std(axis=0, ddof=0)

# Protect against division by zero.
train_std = np.where(train_std == 0, 1.0, train_std)
X_train_scaled = (X_train_raw - train_mean) / train_std

# Apply the same transformation to test data.
X_test_scaled = (X_test_raw - train_mean) / train_std
scaled_df = pd.DataFrame(X_train_scaled, columns=feature_cols)

# Print compact diagnostics for learners.
print("Raw training feature ranges:")
print(train_df[feature_cols].agg(["min", "max"]))
print("\nScaled training means:", np.round(X_train_scaled.mean(axis=0), 3))
print("Scaled training stds:", np.round(X_train_scaled.std(axis=0), 3))
print("Test matrix shape after scaling:", X_test_scaled.shape)

# Plot raw versus scaled feature spread.
fig, axes = plt.subplots(1, 2, figsize=(9, 3))
axes[0].boxplot(X_train_raw, labels=feature_cols)
axes[0].set_title("Raw feature scales")
axes[0].tick_params(axis="x", rotation=25)
axes[1].boxplot(X_train_scaled, labels=feature_cols)
axes[1].set_title("Scaled feature scales")
axes[1].tick_params(axis="x", rotation=25)
plt.tight_layout()
plt.show()

```

### **1.3. Feature Scaling Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1783558840" width="250">



>* Scale features to comparable numerical ranges
>* Balanced matrices improve model training behavior

>* Large feature values can dominate gradient descent
>* Scaling improves convergence and training stability

>* Scale features before fitting the model
>* Apply the same scaling to new data



```
#@title Python Code - Feature Scaling Basics

# Feature scaling balances multiple engineering predictors.
# This example uses pandas and NumPy.
# Scaled columns help gradient descent behave.

# Import libraries for tabular numerical work.
import numpy as np
import pandas as pd

# Create a tiny civil engineering dataset.
data = {
    "floor_area_sqft": [850, 1200, 1500, 1800, 2200, 2600],
    "distance_km": [1.2, 3.5, 2.1, 5.0, 4.2, 6.8],
    "building_age_years": [5, 18, 10, 35, 22, 40],
    "rent_dollars": [1800, 2100, 2500, 2400, 2800, 3000],
}

# Store the data in a DataFrame.
df = pd.DataFrame(data)

# Separate input features from target values.
feature_names = ["floor_area_sqft", "distance_km", "building_age_years"]
X_raw = df[feature_names].to_numpy(dtype=float)
y = df["rent_dollars"].to_numpy(dtype=float)

# Validate the feature matrix dimensions.
assert X_raw.shape == (6, 3)
assert y.shape[0] == X_raw.shape[0]

# Compute training means and standard deviations.
means = X_raw.mean(axis=0)
stds = X_raw.std(axis=0)

# Protect against division by zero.
stds_safe = np.where(stds == 0, 1.0, stds)
X_scaled = (X_raw - means) / stds_safe

# Build a readable scaled feature table.
scaled_df = pd.DataFrame(X_scaled, columns=feature_names)
summary = scaled_df.agg(["mean", "std"]).round(2)

# Show compact evidence of successful scaling.
print("Raw feature ranges:")
print(df[feature_names].agg(["min", "max"]))
print("\nScaled feature summary:")
print(summary)

# Transform one future apartment consistently.
new_home = np.array([[1600.0, 2.8, 12.0]])
new_scaled = (new_home - means) / stds_safe

# Display the scaled future input compactly.
print("\nScaled new input:")
print(np.round(new_scaled, 2))

```

## **2. Gradient Descent Loop**

### **2.1. Vectorized Predictions**

>* Predict all rows in one operation
>* Combine features, weights, and intercept clearly

>* Scales training to large real datasets
>* Speeds repeated gradient descent prediction steps

```
#@title Python Code - Vectorized Predictions

# Vectorized predictions support multi feature regression.
# Each row represents one civil project.
# Each column stores one scaled feature.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
data = pd.DataFrame({
    "area_sqft": [900, 1200, 1500, 1800, 2100, 2400],
    "distance_km": [2.0, 3.5, 5.0, 6.0, 8.0, 9.5],
    "age_years": [5, 10, 15, 20, 25, 30],
    "price_k": [310, 360, 405, 455, 500, 545]
})

# Select multiple input features.
feature_names = ["area_sqft", "distance_km", "age_years"]
X_raw = data[feature_names].to_numpy(dtype=float)
y = data["price_k"].to_numpy(dtype=float)

# Validate matching observation counts.
assert X_raw.shape[0] == y.shape[0]
assert X_raw.shape[1] == len(feature_names)

# Scale features using column statistics.
means = X_raw.mean(axis=0)
stds = X_raw.std(axis=0)
X_scaled = (X_raw - means) / stds

# Add an intercept column of ones.
ones = np.ones((X_scaled.shape[0], 1))
X_design = np.column_stack((ones, X_scaled))
weights = np.array([420.0, 40.0, -10.0, 5.0])

# Compute all predictions in one operation.
predictions = X_design @ weights
errors = predictions - y
mse = np.mean(errors ** 2)

# Show compact vectorized prediction results.
print(f"NumPy version: {np.__version__}")
print(f"Design matrix shape: {X_design.shape}")
print(f"Weights shape: {weights.shape}")
print(f"Predictions shape: {predictions.shape}")
print(f"Mean squared error: {mse:.2f}")

# Plot actual versus vectorized predictions.
plt.figure(figsize=(6, 4))
plt.scatter(y, predictions, color="steelblue", s=70)
plt.plot([300, 560], [300, 560], color="black", linestyle="--")
plt.xlabel("Actual price, thousand dollars")
plt.ylabel("Predicted price, thousand dollars")
plt.title("Vectorized predictions for all projects")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Weight Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1783558846" width="250">



>* Errors guide each feature weight update
>* Systematic updates improve predictions over time

>* Scale features to prevent unfair weight changes
>* Updates reduce error based on feature relationships

>* Bias sets the model’s baseline prediction
>* Repeated updates help errors decrease



```
#@title Python Code - Weight Updates

# This example shows gradient descent weight updates.
# Scaled civil features keep updates balanced.
# Small synthetic data keeps execution fast.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a seed for repeatable results.
rng = np.random.default_rng(7)

# Create tiny housing style training data.
n_samples = 40
area_sqft = rng.normal(1800, 350, n_samples)
bedrooms = rng.integers(2, 6, n_samples)
distance_miles = rng.normal(8, 3, n_samples)

# Build prices from known relationships.
noise = rng.normal(0, 12000, n_samples)
price = 60000 + 120 * area_sqft
price += 18000 * bedrooms - 9000 * distance_miles
price += noise

# Store features in a readable table.
data = pd.DataFrame({
    "area_sqft": area_sqft,
    "bedrooms": bedrooms,
    "distance_miles": distance_miles,
    "price": price,
})

# Separate inputs and target values.
feature_cols = ["area_sqft", "bedrooms", "distance_miles"]
X_raw = data[feature_cols].to_numpy(dtype=float)
y = data["price"].to_numpy(dtype=float)

# Validate shapes before training begins.
assert X_raw.shape == (n_samples, 3)
assert y.shape == (n_samples,)

# Scale each feature using training statistics.
means = X_raw.mean(axis=0)
stds = X_raw.std(axis=0)
stds = np.where(stds == 0, 1, stds)
X = (X_raw - means) / stds

# Initialize weights and bias near zero.
weights = np.zeros(X.shape[1])
bias = 0.0
learning_rate = 0.08
iterations = 120

# Track loss values for visual diagnosis.
loss_history = []

# Run the gradient descent update loop.
for step in range(iterations):
    predictions = X @ weights + bias
    errors = predictions - y
    loss = np.mean(errors ** 2)
    loss_history.append(loss)

    # Compute gradients for all weights.
    weight_gradients = (2 / n_samples) * (X.T @ errors)
    bias_gradient = 2 * errors.mean()

    # Update weights and bias together.
    weights -= learning_rate * weight_gradients
    bias -= learning_rate * bias_gradient

# Compare starting and ending training error.
start_rmse = np.sqrt(loss_history[0])
final_rmse = np.sqrt(loss_history[-1])

# Print concise teaching results only.
print(f"NumPy version: {np.__version__}")
print(f"Initial RMSE: ${start_rmse:,.0f}")
print(f"Final RMSE:   ${final_rmse:,.0f}")
print("Learned scaled weights:")
for name, value in zip(feature_cols, weights):
    print(f"  {name}: {value:,.0f}")
print(f"Learned bias: {bias:,.0f}")

# Plot loss decrease across updates.
plt.figure(figsize=(7, 4))
plt.plot(loss_history, color="steelblue", linewidth=2)
plt.xlabel("Gradient descent iteration")
plt.ylabel("Mean squared error")
plt.title("Weight updates reduce training error")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **2.3. Learning Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1783558842" width="250">



>* Learning rate sets gradient descent step size
>* Balanced steps reduce error stably

>* Multiple features need balanced learning steps
>* Scaling helps weights update smoothly

>* Monitor error trends to tune learning rate
>* Choose rates for steady, reliable progress



```
#@title Python Code - Learning Rate

# Learning rates control gradient descent step sizes.
# Scaled civil features make updates more stable.
# This example compares three learning rates.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny deterministic civil engineering dataset.
rng = np.random.default_rng(7)
area_m2 = rng.uniform(60, 180, 40)

# Add distance and age features.
distance_km = rng.uniform(1, 18, 40)
age_years = rng.uniform(0, 60, 40)

# Build target construction cost values.
noise = rng.normal(0, 8, 40)
cost_k = 80 + 1.4 * area_m2

cost_k = cost_k - 2.2 * distance_km
cost_k = cost_k - 0.7 * age_years + noise

# Store features in a pandas table.
data = pd.DataFrame({"area_m2": area_m2, "distance_km": distance_km, "age_years": age_years})

# Scale features using z-score scaling.
means = data.mean()
stds = data.std(ddof=0)

X_scaled = ((data - means) / stds).to_numpy()
y = cost_k.reshape(-1, 1)

# Add an intercept column for bias.
ones = np.ones((X_scaled.shape[0], 1))
X = np.hstack([ones, X_scaled])

# Validate shapes before gradient descent.
assert X.shape[0] == y.shape[0]
assert X.shape[1] == 4

# Define mean squared error calculation.
def mse_loss(X_matrix, y_vector, weights):
    predictions = X_matrix @ weights
    errors = predictions - y_vector

    return float(np.mean(errors ** 2))

# Train linear regression using gradient descent.
def train_with_rate(rate, steps=80):
    weights = np.zeros((X.shape[1], 1))
    losses = []

    for step in range(steps):
        predictions = X @ weights
        gradient = (2 / len(y)) * X.T @ (predictions - y)

        weights = weights - rate * gradient
        losses.append(mse_loss(X, y, weights))

    return weights, losses

# Compare small, useful, and large learning rates.
rates = [0.001, 0.05, 0.8]
results = {rate: train_with_rate(rate) for rate in rates}

# Print a compact summary of final errors.
print("NumPy version:", np.__version__)
for rate, (_, losses) in results.items():
    print(f"rate={rate}: final MSE={losses[-1]:.2f}")

# Plot loss curves for visual diagnosis.
plt.figure(figsize=(7, 4))
for rate, (_, losses) in results.items():
    plt.plot(losses, label=f"learning rate {rate}")

# Format the diagnostic plot clearly.
plt.xlabel("Gradient descent step")
plt.ylabel("Mean squared error")
plt.title("Learning rate affects training stability")

plt.yscale("log")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

## **3. Comparing Regression Models**

### **3.1. Test Error Comparison**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1783558830" width="250">



>* Test error checks predictions on unseen data
>* Choose models that generalize reliably

>* Compare models using identical test conditions
>* Weigh error gains against model complexity

>* Judge errors by real-world context
>* Check where overall error hides problems



```
#@title Python Code - Test Error Comparison

# Compare regression models using held-out test error.
# Use small synthetic civil engineering data.
# Keep outputs short and beginner friendly.

# NumPy handles arrays and random sampling.
import numpy as np

# Pandas stores the small tabular dataset.
import pandas as pd

# Matplotlib creates one diagnostic comparison plot.
import matplotlib.pyplot as plt

# Set a seed for repeatable results.
rng = np.random.default_rng(42)

# Create a small bridge cost dataset.
n_samples = 80
span_ft = rng.uniform(40, 220, n_samples)
traffic_k = rng.uniform(5, 80, n_samples)
soil_score = rng.uniform(1, 10, n_samples)

# Build costs include signal plus noise.
noise = rng.normal(0, 120, n_samples)
cost_k = 180 + 4.2 * span_ft + 6.5 * traffic_k + 35 * soil_score + noise

# Store features and target together.
data = pd.DataFrame({
    "span_ft": span_ft,
    "traffic_k": traffic_k,
    "soil_score": soil_score,
    "cost_k": cost_k,
})

# Shuffle once, then keep one shared split.
data = data.sample(frac=1, random_state=7).reset_index(drop=True)
train_size = int(0.75 * len(data))
train_df = data.iloc[:train_size].copy()
test_df = data.iloc[train_size:].copy()

# Confirm the split is usable.
assert len(train_df) > 10 and len(test_df) > 5

# Define a small feature scaling helper.
def scale_features(train_values, test_values):
    means = train_values.mean(axis=0)
    stds = train_values.std(axis=0)
    stds = np.where(stds == 0, 1, stds)
    return (train_values - means) / stds, (test_values - means) / stds

# Define gradient descent for linear regression.
def fit_linear_gd(X, y, steps=1200, lr=0.05):
    Xb = np.c_[np.ones(len(X)), X]
    weights = np.zeros(Xb.shape[1])
    for step in range(steps):
        errors = Xb @ weights - y
        gradient = Xb.T @ errors / len(y)
        weights -= lr * gradient
    return weights

# Define prediction and mean absolute error.
def predict_linear(X, weights):
    Xb = np.c_[np.ones(len(X)), X]
    return Xb @ weights

# Define a compact error metric.
def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

# Prepare one-feature and three-feature models.
y_train = train_df["cost_k"].to_numpy()
y_test = test_df["cost_k"].to_numpy()

# Scale using training statistics only.
X1_train, X1_test = scale_features(
    train_df[["span_ft"]].to_numpy(),
    test_df[["span_ft"]].to_numpy(),
)

# Scale all candidate engineering features.
X3_train, X3_test = scale_features(
    train_df[["span_ft", "traffic_k", "soil_score"]].to_numpy(),
    test_df[["span_ft", "traffic_k", "soil_score"]].to_numpy(),
)

# Train both models silently from scratch.
w_one = fit_linear_gd(X1_train, y_train)
w_three = fit_linear_gd(X3_train, y_train)

# Evaluate both models on identical test rows.
pred_one = predict_linear(X1_test, w_one)
pred_three = predict_linear(X3_test, w_three)

# Compute comparable test errors.
mae_one = mean_absolute_error(y_test, pred_one)
mae_three = mean_absolute_error(y_test, pred_three)
improvement = mae_one - mae_three

# Print only a few teaching lines.
print(f"NumPy version: {np.__version__}")
print(f"Shared test examples: {len(y_test)}")
print(f"One-feature test MAE: ${mae_one:,.1f}k")
print(f"Three-feature test MAE: ${mae_three:,.1f}k")
print(f"MAE improvement: ${improvement:,.1f}k")

# Plot predicted versus actual test costs.
plt.figure(figsize=(6, 4))
plt.scatter(y_test, pred_one, label="One feature", alpha=0.75)
plt.scatter(y_test, pred_three, label="Three features", alpha=0.75)

# Add a perfect-prediction reference line.
low = min(y_test.min(), pred_one.min(), pred_three.min())
high = max(y_test.max(), pred_one.max(), pred_three.max())
plt.plot([low, high], [low, high], "k--", label="Perfect")

# Label the diagnostic plot clearly.
plt.xlabel("Actual bridge cost, thousand dollars")
plt.ylabel("Predicted bridge cost, thousand dollars")
plt.title("Test Error Comparison on Same Held-Out Bridges")
plt.legend()

# Display the single visual diagnostic.
plt.tight_layout()
plt.show()

```

### **3.2. Prediction Diagnostics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1783558832" width="250">

>* Check error patterns for trustworthy predictions



```
#@title Python Code - Prediction Diagnostics

# Compare models using prediction diagnostics.
# Residual patterns reveal hidden weaknesses.
# Civil datasets often need careful checking.

# Import numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set deterministic random seed.
rng = np.random.default_rng(42)

# Create a small bridge inspection dataset.
n = 80
age_years = rng.uniform(1, 60, n)
traffic_thousands = rng.uniform(2, 45, n)
span_meters = rng.uniform(10, 120, n)

# Build nonlinear deterioration target values.
noise = rng.normal(0, 3.5, n)
condition_score = 95 - 0.55 * age_years
condition_score -= 0.18 * traffic_thousands
condition_score -= 0.0018 * span_meters ** 2
condition_score += noise

# Store features and target together.
data = pd.DataFrame({
    "age_years": age_years,
    "traffic_thousands": traffic_thousands,
    "span_meters": span_meters,
    "condition_score": condition_score,
})

# Split rows into train and test sets.
indices = rng.permutation(n)
train_idx = indices[:60]
test_idx = indices[60:]

# Prepare two competing feature sets.
features_simple = ["age_years", "traffic_thousands"]
features_richer = ["age_years", "traffic_thousands", "span_meters"]

# Define a safe scaling helper.
def scale_train_test(train_frame, test_frame, columns):
    train_values = train_frame[columns].to_numpy()
    test_values = test_frame[columns].to_numpy()
    means = train_values.mean(axis=0)
    stds = train_values.std(axis=0)

    # Prevent division by zero.
    stds = np.where(stds == 0, 1, stds)
    train_scaled = (train_values - means) / stds
    test_scaled = (test_values - means) / stds
    return train_scaled, test_scaled

# Define gradient descent linear regression.
def fit_linear_gd(X, y, steps=1200, rate=0.04):
    X_design = np.c_[np.ones(X.shape[0]), X]
    weights = np.zeros(X_design.shape[1])
    assert X_design.shape[0] == y.shape[0]

    # Update weights silently for teaching clarity.
    for step in range(steps):
        errors = X_design @ weights - y
        gradient = X_design.T @ errors / y.size
        weights -= rate * gradient
    return weights

# Define prediction and error helpers.
def predict_linear(X, weights):
    X_design = np.c_[np.ones(X.shape[0]), X]
    return X_design @ weights

# Train and evaluate one model.
def evaluate_model(feature_names):
    train_frame = data.iloc[train_idx]
    test_frame = data.iloc[test_idx]
    X_train, X_test = scale_train_test(train_frame, test_frame, feature_names)
    y_train = train_frame["condition_score"].to_numpy()

    # Fit model and compute diagnostics.
    y_test = test_frame["condition_score"].to_numpy()
    weights = fit_linear_gd(X_train, y_train)
    predictions = predict_linear(X_test, weights)
    residuals = y_test - predictions
    mae = np.mean(np.abs(residuals))
    return predictions, residuals, mae

# Evaluate both competing models.
pred_simple, res_simple, mae_simple = evaluate_model(features_simple)
pred_richer, res_richer, mae_richer = evaluate_model(features_richer)

# Print compact diagnostic summary.
print(f"NumPy version: {np.__version__}")
print(f"Simple model test MAE: {mae_simple:.2f} score points")
print(f"Richer model test MAE: {mae_richer:.2f} score points")
print(f"Largest simple residual: {np.max(np.abs(res_simple)):.2f}")
print(f"Largest richer residual: {np.max(np.abs(res_richer)):.2f}")

# Plot residuals against actual condition scores.
y_test_plot = data.iloc[test_idx]["condition_score"].to_numpy()
plt.figure(figsize=(7, 4))
plt.axhline(0, color="black", linewidth=1)
plt.scatter(y_test_plot, res_simple, label="Simple model", alpha=0.75)
plt.scatter(y_test_plot, res_richer, label="Richer model", alpha=0.75)

# Label the diagnostic plot clearly.
plt.xlabel("Actual bridge condition score")
plt.ylabel("Residual: actual minus predicted")
plt.title("Prediction Diagnostics on Test Cases")
plt.legend()
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Visual Model Diagnostics

# Visual diagnostics reveal model behavior.
# We compare two regression models.
# Residual plots expose hidden patterns.

# Import small scientific Python libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set deterministic randomness for reproducibility.
rng = np.random.default_rng(42)
n_samples = 90
indices = np.arange(n_samples)

# Create civil engineering style features.
area_sqft = rng.uniform(900, 3200, n_samples)
age_years = rng.uniform(0, 80, n_samples)
distance_miles = rng.uniform(0.5, 18, n_samples)

# Build a nonlinear target with noise.
noise = rng.normal(0, 18000, n_samples)
price = 65000 + 145 * area_sqft
price += -900 * age_years - 2600 * distance_miles
price += 0.018 * area_sqft * age_years + noise

# Store data in a compact dataframe.
data = pd.DataFrame({
    "area_sqft": area_sqft,
    "age_years": age_years,
    "distance_miles": distance_miles,
    "price": price,
})

# Split rows into train and test sets.
rng.shuffle(indices)
train_idx = indices[:65]
test_idx = indices[65:]
train = data.iloc[train_idx].copy()
test = data.iloc[test_idx].copy()

# Define feature sets for comparison.
basic_cols = ["area_sqft", "age_years"]
full_cols = ["area_sqft", "age_years", "distance_miles"]

# Scale features using training statistics only.
def scaled_matrix(frame, columns, means, stds):
    values = frame[columns].to_numpy(dtype=float)
    scaled = (values - means) / stds
    ones = np.ones((len(frame), 1))
    return np.column_stack([ones, scaled])

# Fit linear regression using normal equations.
def fit_linear(train_frame, columns):
    means = train_frame[columns].mean().to_numpy()
    stds = train_frame[columns].std(ddof=0).to_numpy()
    stds = np.where(stds == 0, 1, stds)
    X_train = scaled_matrix(train_frame, columns, means, stds)
    y_train = train_frame["price"].to_numpy(dtype=float)
    weights = np.linalg.pinv(X_train.T @ X_train) @ X_train.T @ y_train
    return weights, means, stds

# Predict prices for any dataframe.
def predict(frame, columns, weights, means, stds):
    X_values = scaled_matrix(frame, columns, means, stds)
    return X_values @ weights

# Train both candidate regression models.
basic_w, basic_m, basic_s = fit_linear(train, basic_cols)
full_w, full_m, full_s = fit_linear(train, full_cols)

# Generate test predictions for diagnostics.
y_test = test["price"].to_numpy(dtype=float)
basic_pred = predict(test, basic_cols, basic_w, basic_m, basic_s)
full_pred = predict(test, full_cols, full_w, full_m, full_s)

# Validate prediction shapes before evaluation.
assert basic_pred.shape == y_test.shape
assert full_pred.shape == y_test.shape

# Compute test root mean squared errors.
basic_rmse = np.sqrt(np.mean((y_test - basic_pred) ** 2))
full_rmse = np.sqrt(np.mean((y_test - full_pred) ** 2))

# Print concise comparison results.
print(f"NumPy version: {np.__version__}")
print(f"Basic model test RMSE: ${basic_rmse:,.0f}")
print(f"Full model test RMSE:  ${full_rmse:,.0f}")
print("Plot: actual versus predicted, plus residuals.")

# Prepare residuals for visual diagnostics.
basic_resid = y_test - basic_pred
full_resid = y_test - full_pred
low = min(y_test.min(), basic_pred.min(), full_pred.min())
high = max(y_test.max(), basic_pred.max(), full_pred.max())

# Create exactly one diagnostic figure.
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].scatter(y_test, basic_pred, label="Basic", alpha=0.75)
axes[0].scatter(y_test, full_pred, label="Full", alpha=0.75)
axes[0].plot([low, high], [low, high], "k--", linewidth=1)

# Label the actual versus predicted panel.
axes[0].set_title("Actual vs Predicted")
axes[0].set_xlabel("Actual price")
axes[0].set_ylabel("Predicted price")
axes[0].legend()

# Plot residuals against predicted values.
axes[1].scatter(basic_pred, basic_resid, label="Basic", alpha=0.75)
axes[1].scatter(full_pred, full_resid, label="Full", alpha=0.75)
axes[1].axhline(0, color="black", linestyle="--", linewidth=1)

# Label the residual diagnostic panel.
axes[1].set_title("Residuals vs Predicted")
axes[1].set_xlabel("Predicted price")
axes[1].set_ylabel("Actual minus predicted")
axes[1].legend()

# Display the single diagnostic figure.
fig.tight_layout()
plt.show()

```
