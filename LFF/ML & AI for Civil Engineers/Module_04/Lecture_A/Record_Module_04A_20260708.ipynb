## **1. Linear Model Equation**

### **1.1. Model Weight**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1783557674" width="250">



>* Weight shows feature influence strength
>* Sign shows prediction increase or decrease

>* Weights are learned from training data
>* Interpret weights using feature units and scale

>* Weight shows prediction change rate
>* Association, not guaranteed causation



```
#@title Python Code - Model Weight

# This example shows model weight meaning.
# Floor area predicts project construction cost.
# Different weights create different slopes.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
area_m2 = np.array([80, 120, 160, 200, 240], dtype=float)
cost_kusd = np.array([180, 250, 330, 390, 470], dtype=float)

# Validate matching feature and target lengths.
assert area_m2.shape == cost_kusd.shape
assert area_m2.size >= 2

# Define two possible model weights.
small_weight = 1.2
large_weight = 1.8

# Use one shared bias value.
bias = 70.0
pred_small = small_weight * area_m2 + bias
pred_large = large_weight * area_m2 + bias

# Print compact interpretation lines.
print("Feature: floor area in square meters")
print("Target: construction cost in thousand dollars")
print(f"Small weight {small_weight}: gentler cost increase")
print(f"Large weight {large_weight}: steeper cost increase")
print("Weight is the slope of each prediction line")

# Plot data and two prediction lines.
plt.figure(figsize=(7, 4))
plt.scatter(area_m2, cost_kusd, label="Observed projects")
plt.plot(area_m2, pred_small, label="Weight = 1.2")

# Add the steeper model line.
plt.plot(area_m2, pred_large, label="Weight = 1.8")
plt.xlabel("Floor area (m²)")
plt.ylabel("Predicted cost (thousand dollars)")
plt.title("Model Weight Controls Linear Regression Slope")

# Display legend and grid clearly.
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **1.2. Feature Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1783557675" width="250">



>* Features provide measured inputs for predictions
>* Weights and bias turn features into outputs

>* Choose features related to the prediction target
>* Reliable features improve model prediction quality

>* Feature meaning depends on scale and context
>* Check whether linear assumptions remain reasonable



```
#@title Python Code - Feature Inputs

# Feature inputs are measured model information.
# Linear regression combines features with parameters.
# Civil examples make feature meaning concrete.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature: beam span length.
span_meters = np.array([3, 4, 5, 6, 7], dtype=float)

# Store matching target deflections.
deflection_mm = np.array([5.2, 7.1, 8.9, 11.0, 12.8], dtype=float)

# Validate paired feature and target values.
assert span_meters.shape == deflection_mm.shape
assert span_meters.size >= 2

# Choose simple linear model parameters.
weight = 1.9
bias = -0.5

# Compute predictions from feature values.
predicted_mm = weight * span_meters + bias

# Compute residuals as target minus prediction.
residuals_mm = deflection_mm - predicted_mm

# Print compact teaching results.
print("Feature input: beam span length in meters")
print("Target output: measured deflection in millimeters")
print(f"Model equation: prediction = {weight} * span + {bias}")
print(f"For span {span_meters[2]:.1f} m, prediction is {predicted_mm[2]:.1f} mm")
print(f"Residual there is {residuals_mm[2]:.1f} mm")

# Plot feature values against targets and predictions.
plt.figure(figsize=(6, 4))
plt.scatter(span_meters, deflection_mm, label="Measured targets")
plt.plot(span_meters, predicted_mm, label="Linear predictions")
plt.xlabel("Feature input: beam span length (m)")
plt.ylabel("Target: deflection (mm)")
plt.title("One Feature Input Drives Linear Predictions")
plt.legend()
plt.grid(True)
plt.show()

```

### **1.3. Target Prediction**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1783557672" width="250">



>* Inputs combine with weights to predict targets
>* Models generalize patterns to new cases

>* Predictions are estimates, not guaranteed outcomes
>* Features combine with weights and bias

>* Simple predictions are easy to trace
>* Unexpected results reveal input relationships



```
#@title Python Code - Target Prediction

# Linear regression predicts targets from features.
# We use concrete strength as example.
# Weight and bias create predictions.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature in centimeters.
curing_days = np.array([3, 7, 14, 21, 28], dtype=float)

# Store observed target strengths.
strength_mpa = np.array([12, 20, 29, 34, 38], dtype=float)

# Validate matching feature and target lengths.
assert curing_days.shape == strength_mpa.shape

# Choose an interpretable weight and bias.
weight = 0.95
bias = 11.0

# Predict targets using the linear equation.
predicted_mpa = weight * curing_days + bias

# Compute residuals as observed minus predicted.
residuals_mpa = strength_mpa - predicted_mpa

# Print the model equation clearly.
print(f"Prediction equation: strength = {weight:.2f} * days + {bias:.1f}")

# Print one new target prediction.
new_days = 18.0
new_prediction = weight * new_days + bias
print(f"For {new_days:.0f} curing days, predicted strength is {new_prediction:.1f} MPa")

# Print compact residual information.
print(f"Residuals show errors: {np.round(residuals_mpa, 1).tolist()} MPa")

# Create one plot comparing observations and predictions.
plt.figure(figsize=(7, 4))
plt.scatter(curing_days, strength_mpa, label="Observed targets")
plt.plot(curing_days, predicted_mpa, color="orange", label="Linear predictions")
plt.scatter([new_days], [new_prediction], color="red", label="New prediction")
plt.xlabel("Curing time, days")
plt.ylabel("Concrete strength, MPa")
plt.title("Target Prediction With Weight and Bias")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

## **2. Loss and Descent**

### **2.1. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1783557662" width="250">



>* Measures average squared prediction errors
>* Penalizes large mistakes more strongly

>* MSE gives models a clear improvement target
>* Lower MSE means better overall predictions

>* MSE guides gradient descent toward better lines
>* Low training error may not generalize



```
#@title Python Code - Mean Squared Error

# This example computes mean squared error.
# Bridge age predicts maintenance cost.
# Gradient descent improves one regression line.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Store one feature and target values.
age_years = np.array([5, 10, 15, 20, 25, 30], dtype=float)
cost_kusd = np.array([18, 24, 31, 39, 45, 53], dtype=float)

# Validate matching one dimensional data.
assert age_years.ndim == cost_kusd.ndim == 1
assert age_years.size == cost_kusd.size

# Define prediction and loss functions.
def predict(x_values, weight, bias):
    return weight * x_values + bias

# Average squared residuals measure error.
def mean_squared_error(actual, predicted):
    residuals = predicted - actual
    return np.mean(residuals ** 2)

# Start with a simple initial line.
weight = 0.0
bias = 0.0
learning_rate = 0.002

# Run a short silent descent loop.
for step in range(800):
    predictions = predict(age_years, weight, bias)
    errors = predictions - cost_kusd
    weight -= learning_rate * np.mean(2 * errors * age_years)

    bias -= learning_rate * np.mean(2 * errors)

# Compute final predictions and residuals.
final_predictions = predict(age_years, weight, bias)
final_mse = mean_squared_error(cost_kusd, final_predictions)
residuals = cost_kusd - final_predictions

# Print a compact interpretation summary.
print(f"Slope: {weight:.2f} thousand dollars per year")
print(f"Intercept: {bias:.2f} thousand dollars")
print(f"Mean squared error: {final_mse:.2f}")
print(f"First residual: {residuals[0]:.2f} thousand dollars")

# Plot observations and fitted regression line.
plt.figure(figsize=(6, 4))
plt.scatter(age_years, cost_kusd, label="Observed bridges")
plt.plot(age_years, final_predictions, color="red", label="Fitted line")

# Label the civil engineering axes.
plt.xlabel("Bridge age (years)")
plt.ylabel("Maintenance cost (thousand dollars)")
plt.title("Mean Squared Error for One-Feature Regression")

# Show legend, grid, and plot.
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Gradient Descent**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1783557664" width="250">



>* Adjusts slope and intercept to reduce error
>* Moves the line toward better data fit

>* Loss surface guides parameter choices
>* Gradients update slope and intercept

>* Repeatedly update slope and intercept
>* Learn overall feature-target relationship



```
#@title Python Code - Gradient Descent

# Gradient descent adjusts regression line parameters.
# Civil engineering data keeps examples concrete.
# Small arrays make each calculation visible.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Store pavement thickness and load capacity.
thickness_in = np.array([4, 5, 6, 7, 8, 9], dtype=float)
capacity_kips = np.array([18, 23, 27, 31, 36, 40], dtype=float)

# Validate matching one dimensional training arrays.
assert thickness_in.ndim == capacity_kips.ndim == 1
assert thickness_in.size == capacity_kips.size

# Scale feature values for stable descent.
x_mean = thickness_in.mean()
x_std = thickness_in.std()

# Create standardized one feature input.
x_scaled = (thickness_in - x_mean) / x_std
y_target = capacity_kips.copy()

# Define mean squared error function.
def mean_squared_error(y_true, y_pred):
    errors = y_pred - y_true
    return np.mean(errors ** 2)

# Initialize slope, intercept, and learning rate.
slope = 0.0
intercept = y_target.mean()
learning_rate = 0.08

# Run a short silent gradient descent loop.
loss_history = []
for step in range(120):
    predictions = slope * x_scaled + intercept
    residuals = predictions - y_target

    slope_gradient = 2 * np.mean(residuals * x_scaled)
    intercept_gradient = 2 * np.mean(residuals)
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient

    current_loss = mean_squared_error(y_target, predictions)
    loss_history.append(current_loss)

# Convert scaled slope back to original units.
original_slope = slope / x_std
original_intercept = intercept - original_slope * x_mean

# Predict capacities using learned original line.
learned_predictions = original_slope * thickness_in + original_intercept
final_loss = mean_squared_error(capacity_kips, learned_predictions)

# Print concise interpretation for learners.
print(f"Initial MSE: {loss_history[0]:.2f} kip^2")
print(f"Final MSE: {final_loss:.2f} kip^2")
print(f"Slope: {original_slope:.2f} kips per inch")
print(f"Intercept: {original_intercept:.2f} kips")
print(f"Residual at 7 inches: {learned_predictions[3] - capacity_kips[3]:.2f} kips")

# Plot data points and learned regression line.
plt.figure(figsize=(7, 4))
plt.scatter(thickness_in, capacity_kips, label="Observed data")
plt.plot(thickness_in, learned_predictions, color="red", label="Learned line")

# Label the civil engineering regression plot.
plt.xlabel("Pavement thickness (inches)")
plt.ylabel("Load capacity (kips)")
plt.title("Gradient Descent Learns Slope and Intercept")
plt.legend()

# Display the single teaching plot.
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Learning Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1783557661" width="250">



>* Learning rate controls gradient descent step size
>* Too small slows; too large destabilizes

>* Good rates steadily reduce prediction error
>* Monitor loss to avoid unstable updates

>* Learning rate balances speed and stability
>* Smooth loss decline shows effective learning



```
#@title Python Code - Learning Rate

# Learning rate controls descent step size.
# Concrete strength data supports regression learning.
# Small examples reveal stable training behavior.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature and target values.
cement_kg = np.array([250, 280, 310, 340, 370], dtype=float)
strength_mpa = np.array([24, 28, 31, 35, 39], dtype=float)

# Validate matching one dimensional arrays.
assert cement_kg.ndim == strength_mpa.ndim == 1
assert cement_kg.size == strength_mpa.size

# Scale cement values for stable descent.
x_mean = cement_kg.mean()
x_std = cement_kg.std()
x_scaled = (cement_kg - x_mean) / x_std

# Define mean squared error calculation.
def mse_loss(weight, bias, x_values, y_values):
    predictions = weight * x_values + bias
    errors = predictions - y_values
    return np.mean(errors ** 2)

# Run gradient descent for one learning rate.
def train_with_rate(rate, steps=40):
    weight = 0.0
    bias = 0.0
    losses = []

    for step in range(steps):
        predictions = weight * x_scaled + bias
        errors = predictions - strength_mpa
        losses.append(mse_loss(weight, bias, x_scaled, strength_mpa))

        weight_gradient = 2 * np.mean(errors * x_scaled)
        bias_gradient = 2 * np.mean(errors)
        weight = weight - rate * weight_gradient
        bias = bias - rate * bias_gradient

    return weight, bias, losses

# Compare cautious, moderate, and aggressive rates.
rates = [0.01, 0.10, 0.80]
results = {rate: train_with_rate(rate) for rate in rates}

# Print a compact summary of learning behavior.
print("Learning rate comparison for concrete strength regression")
for rate, (weight, bias, losses) in results.items():
    print(f"rate={rate:.2f}, start MSE={losses[0]:.1f}, final MSE={losses[-1]:.1f}")

# Plot loss curves to show descent stability.
plt.figure(figsize=(7, 4))
for rate, (_, _, losses) in results.items():
    plt.plot(losses, marker="o", markersize=3, label=f"rate {rate}")

# Label the plot for interpretation.
plt.xlabel("Gradient descent step")
plt.ylabel("Mean squared error")
plt.title("Learning rate changes the descent path")
plt.legend()

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```

## **3. Engineering Meaning**

### **3.1. Slope Meaning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1783557668" width="250">



>* Slope shows predicted change per input unit
>* It expresses engineering sensitivity and rate

>* Slope sign shows relationship direction
>* Slope size shows prediction sensitivity

>* State slope units clearly
>* Interpret only within engineering data limits



```
#@title Python Code - Slope Meaning

# This example interprets slope in engineering.
# Road length predicts pavement repair cost.
# Units make the slope meaningful.

import numpy as np
import matplotlib.pyplot as plt

# Store road lengths in kilometers.
length_km = np.array([1, 2, 3, 4, 5], dtype=float)

# Store repair costs in thousand dollars.
cost_k = np.array([42, 55, 69, 82, 96], dtype=float)

# Validate matching one dimensional data.
assert length_km.ndim == cost_k.ndim == 1
assert length_km.size == cost_k.size

# Fit a straight line manually.
slope, intercept = np.polyfit(length_km, cost_k, 1)

# Compute predictions and residuals.
predicted_k = slope * length_km + intercept
residuals_k = cost_k - predicted_k

# Print the engineering interpretation.
print(f"Slope: {slope:.2f} thousand dollars per kilometer")
print(f"Intercept: {intercept:.2f} thousand dollars at zero kilometers")
print(f"Meaning: each extra kilometer adds about ${slope:.2f}k")
print(f"Residual at 3 km: {residuals_k[2]:.2f} thousand dollars")

# Plot data and fitted line.
plt.figure(figsize=(6, 4))
plt.scatter(length_km, cost_k, label="Observed projects")
plt.plot(length_km, predicted_k, label="Linear prediction")
plt.xlabel("Road length (km)")
plt.ylabel("Repair cost (thousand dollars)")
plt.title("Slope as cost increase per kilometer")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.2. Residual Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1783557666" width="250">



>* Residuals show prediction error size
>* Errors reveal useful engineering clues

>* Small residuals show close model predictions
>* Large residuals reveal missing engineering factors

>* Random residuals suggest more reliable models
>* Patterns reveal uncertainty and model limits



```
#@title Python Code - Residual Interpretation

# Residuals compare predictions with observed engineering measurements.
# Signs show whether models overpredict or underpredict.
# Patterns reveal missing variables or unusual cases.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Store water cement ratios.
water_cement_ratio = np.array([0.38, 0.42, 0.46, 0.50, 0.54, 0.58])

# Store observed concrete strengths.
observed_strength_mpa = np.array([43.0, 39.0, 35.0, 31.0, 28.0, 24.0])

# Define a simple fitted line.
slope_mpa_per_ratio = -95.0
intercept_mpa = 79.0

# Compute model predictions.
predicted_strength_mpa = intercept_mpa + slope_mpa_per_ratio * water_cement_ratio

# Validate matching vector lengths.
assert observed_strength_mpa.shape == predicted_strength_mpa.shape

# Compute residuals as observed minus predicted.
residuals_mpa = observed_strength_mpa - predicted_strength_mpa

# Find the largest absolute residual.
largest_index = int(np.argmax(np.abs(residuals_mpa)))

# Summarize the fitted engineering meaning.
print(f"Slope: {slope_mpa_per_ratio:.1f} MPa per ratio unit")
print(f"Intercept: {intercept_mpa:.1f} MPa at zero ratio")
print(f"Mean residual: {np.mean(residuals_mpa):.2f} MPa")
print(f"Largest residual: {residuals_mpa[largest_index]:.2f} MPa")
print(f"Case ratio: {water_cement_ratio[largest_index]:.2f}")

# Interpret the largest residual sign.
if residuals_mpa[largest_index] > 0:
    print("Observed strength exceeded the model prediction.")
else:
    print("Observed strength fell below the model prediction.")

# Create one residual plot.
plt.figure(figsize=(7, 4))
plt.axhline(0, color="black", linewidth=1)
plt.scatter(predicted_strength_mpa, residuals_mpa, color="tab:blue", s=70)
plt.xlabel("Predicted compressive strength, MPa")
plt.ylabel("Residual, observed minus predicted, MPa")
plt.title("Residuals for concrete strength predictions")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Prediction Residuals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1783557670" width="250">



>* Residuals compare predictions with observed results
>* They reveal conservative or optimistic model errors

>* Check residual size, direction, and patterns
>* Use residuals to diagnose missing factors

>* Residual patterns test model usefulness
>* Biased errors reveal missing real-world factors



```
#@title Python Code - Prediction Residuals

# Residuals compare predictions with observed engineering measurements.
# Positive residuals mean observations exceed predictions.
# Patterns in residuals reveal missing engineering factors.

import numpy as np
import matplotlib.pyplot as plt

# Store curing time and measured concrete strength.
days = np.array([3, 7, 14, 21, 28, 35], dtype=float)
strength = np.array([18, 25, 33, 38, 43, 45], dtype=float)

# Define a simple fitted engineering model.
slope = 0.82
intercept = 17.0

# Compute predictions and residuals.
predicted = intercept + slope * days
residuals = strength - predicted

# Check matching data lengths before interpretation.
assert days.shape == strength.shape == residuals.shape

# Summarize residual meaning with limited output.
print("Model: strength = 17.0 + 0.82 * curing_days")
print("Residual = observed strength - predicted strength")
print("Positive residual: model underestimated measured strength")
print("Negative residual: model overestimated measured strength")

# Print compact residual table for learners.
for day, obs, pred, res in zip(days, strength, predicted, residuals):
    print(f"Day {day:>4.0f}: observed {obs:>5.1f}, predicted {pred:>5.1f}, residual {res:>5.1f} MPa")

# Create one residual plot for diagnosis.
plt.figure(figsize=(7, 4))
plt.axhline(0, color="black", linewidth=1)
plt.scatter(days, residuals, color="steelblue", s=70)

# Label axes using engineering units.
plt.xlabel("Curing time, days")
plt.ylabel("Residual, MPa")
plt.title("Concrete strength prediction residuals")

# Add grid and display the plot.
plt.grid(True, alpha=0.3)
plt.show()

```
