## **1. Linear Model Equation**

### **1.1. Weights**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1783738874" width="250">



>* Weights show each feature’s prediction influence
>* Sign and size indicate direction and strength

>* Weights are learned by comparing predictions
>* Weight size depends on feature scale

>* Weights help explain model predictions
>* Interpret weights as associations, not causes



```
#@title Python Code - Weights

# This example shows weights in linear regression.
# Slab thickness predicts load capacity here.
# We compare different possible weight values.

import numpy as np
import matplotlib.pyplot as plt

# Store one civil engineering feature.
thickness_cm = np.array([10, 12, 14, 16, 18, 20])

# Store known target load capacities.
capacity_kn = np.array([55, 63, 72, 80, 89, 97])

# Validate matching feature and target lengths.
assert thickness_cm.shape == capacity_kn.shape

# Choose an intercept for simple comparison.
bias_kn = 5

# Try three possible weight values.
weights = np.array([3.0, 4.0, 5.0])

# Compute predictions for each weight.
predictions = bias_kn + weights[:, None] * thickness_cm

# Compute mean squared error for each weight.
errors = ((predictions - capacity_kn) ** 2).mean(axis=1)

# Find the best weight among candidates.
best_index = int(np.argmin(errors))

# Print a compact interpretation table.
print("Weight means predicted kN gained per centimeter.")
print(f"Bias: {bias_kn} kN when thickness is zero.")

# Print each candidate weight result.
for weight, mse in zip(weights, errors):
    print(f"weight={weight:.1f}, MSE={mse:.2f}")

# Print the best candidate weight.
print(f"Best tested weight: {weights[best_index]:.1f} kN/cm")

# Plot observed data and best line.
plt.figure(figsize=(6, 4))
plt.scatter(thickness_cm, capacity_kn, label="Observed targets")

# Add the best prediction line.
plt.plot(thickness_cm, predictions[best_index], label="Best weighted line")
plt.xlabel("Slab thickness (cm)")
plt.ylabel("Load capacity (kN)")

# Add a clear title and legend.
plt.title("Weight controls the slope of predictions")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Feature Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1783738876" width="250">



>* Features describe each case being predicted
>* Models combine features with weights and bias

>* Choose relevant, measurable, consistent features
>* Consistent definitions prevent misleading predictions

>* Features guide how predictions change
>* Inputs inform, but are not predictions



```
#@title Python Code - Feature Inputs

# Feature inputs describe each engineering observation.
# Linear models combine features with weights.
# This example predicts pavement repair cost.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature using consistent units.
pavement_age_years = np.array([2, 4, 6, 8, 10], dtype=float)

# Store target values measured in thousands.
repair_cost_thousands = np.array([18, 25, 31, 39, 46], dtype=float)

# Choose a simple learned slope and intercept.
slope_per_year = 3.5
intercept_thousands = 10.0

# Compute predictions from feature inputs.
predicted_costs = slope_per_year * pavement_age_years + intercept_thousands

# Compute residuals as target minus prediction.
residuals = repair_cost_thousands - predicted_costs

# Print a compact explanation of feature inputs.
print("Feature: pavement age in years")
print("Target: repair cost in thousand dollars")
print(f"Model: cost = {slope_per_year} * age + {intercept_thousands}")
print(f"For age {pavement_age_years[2]:.0f}, prediction is {predicted_costs[2]:.1f}")
print(f"Actual target is {repair_cost_thousands[2]:.1f}, residual is {residuals[2]:.1f}")

# Plot feature inputs against targets and predictions.
plt.figure(figsize=(6, 4))
plt.scatter(pavement_age_years, repair_cost_thousands, label="targets")
plt.plot(pavement_age_years, predicted_costs, label="linear prediction")
plt.xlabel("Pavement age feature, years")
plt.ylabel("Repair cost target, thousand dollars")
plt.title("Feature Inputs Support Linear Predictions")
plt.legend()
plt.grid(True)
plt.show()

```

### **1.3. Prediction Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1783738878" width="250">



>* Model combines features, weights, and bias
>* Output estimates a target for decisions

>* Predictions are estimates, not certain facts
>* Real-world variation causes prediction errors

>* Connect predictions to target context
>* Use transparent estimates with professional judgment



```
#@title Python Code - Prediction Output

# Linear regression turns features into predictions.
# We use one civil engineering example.
# Outputs are estimates, not guaranteed truths.

import numpy as np
import matplotlib.pyplot as plt

# Store pavement thickness values in inches.
thickness_inches = np.array([4, 5, 6, 7, 8], dtype=float)

# Store observed service life values in years.
actual_years = np.array([9, 11, 13, 14, 17], dtype=float)

# Define a learned slope and intercept.
weight_years_per_inch = 1.9
bias_years = 1.2

# Compute prediction outputs using the linear equation.
predicted_years = weight_years_per_inch * thickness_inches + bias_years

# Validate matching feature and target lengths.
assert thickness_inches.shape == actual_years.shape == predicted_years.shape

# Compute residuals as actual minus predicted values.
residuals = actual_years - predicted_years

# Print the model equation and interpretation.
print(f"Prediction rule: y_hat = {weight_years_per_inch}x + {bias_years}")
print("Feature x: pavement thickness in inches")
print("Target y: pavement service life in years")

# Print one clear prediction example.
example_thickness = 6.5
example_prediction = weight_years_per_inch * example_thickness + bias_years
print(f"For {example_thickness} inches, predicted life is {example_prediction:.1f} years")

# Print compact residual information.
print(f"Residuals mean: {residuals.mean():.2f} years")
print(f"Largest absolute residual: {np.abs(residuals).max():.2f} years")

# Create one plot comparing observations and predictions.
plt.figure(figsize=(7, 4))
plt.scatter(thickness_inches, actual_years, label="Observed targets")
plt.plot(thickness_inches, predicted_years, color="red", label="Prediction output")
plt.xlabel("Pavement thickness (inches)")
plt.ylabel("Service life (years)")
plt.title("Linear Regression Prediction Output")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

## **2. Loss and Descent**

### **2.1. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1783738886" width="250">



>* Measures prediction errors against actual values
>* Lower MSE means a better-fitting line

>* Squaring prevents error cancellation
>* Large mistakes receive stronger penalties

>* MSE measures line fitting errors
>* Gradient descent uses MSE to improve



```
#@title Python Code - Mean Squared Error

# Mean squared error measures prediction mistakes.
# Bridge age predicts yearly maintenance cost.
# We compare two simple regression lines.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature and target values.
ages_years = np.array([5, 10, 15, 20, 25], dtype=float)
costs_kusd = np.array([18, 24, 31, 39, 47], dtype=float)

# Validate matching one dimensional arrays.
assert ages_years.ndim == costs_kusd.ndim == 1
assert ages_years.size == costs_kusd.size

# Define mean squared error directly.
def mean_squared_error(actual_values, predicted_values):
    errors = predicted_values - actual_values
    squared_errors = errors ** 2
    return np.mean(squared_errors)

# Compare two candidate regression lines.
slope_a, intercept_a = 1.0, 12.0
slope_b, intercept_b = 1.45, 9.5

# Compute predictions from slope and intercept.
pred_a = slope_a * ages_years + intercept_a
pred_b = slope_b * ages_years + intercept_b

# Calculate losses for both lines.
mse_a = mean_squared_error(costs_kusd, pred_a)
mse_b = mean_squared_error(costs_kusd, pred_b)

# Print a compact interpretation.
print(f"Line A MSE: {mse_a:.2f} thousand dollars squared")
print(f"Line B MSE: {mse_b:.2f} thousand dollars squared")
print("Smaller MSE means predictions are closer overall.")

# Plot observations and candidate lines.
plt.figure(figsize=(7, 4))
plt.scatter(ages_years, costs_kusd, label="Observed bridge data")

# Draw both prediction lines.
plt.plot(ages_years, pred_a, label="Line A")
plt.plot(ages_years, pred_b, label="Line B")

# Label the engineering context clearly.
plt.xlabel("Bridge age (years)")
plt.ylabel("Maintenance cost (thousand dollars)")
plt.title("Mean Squared Error Compares Regression Lines")

# Show legend and grid.
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Gradient Descent**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1783738888" width="250">



>* Adjust slope and intercept to reduce error
>* Step downhill on the loss surface

>* Residuals guide slope and intercept updates
>* Repeated updates reduce prediction error

>* Improves slope and intercept step by step
>* Reduces loss while still needing good data



```
#@title Python Code - Gradient Descent

# Gradient descent improves a simple regression line.
# Concrete strength data supports civil engineering interpretation.
# Small updates reduce mean squared error.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Store curing time in days.
days = np.array([1, 3, 7, 14, 28, 56], dtype=float)

# Store measured strength in megapascals.
strength = np.array([8, 15, 24, 32, 41, 48], dtype=float)

# Validate matching one dimensional arrays.
assert days.ndim == strength.ndim == 1
assert days.size == strength.size

# Scale input for stable updates.
x_mean = days.mean()
x_std = days.std()

# Create standardized feature values.
x = (days - x_mean) / x_std
n = x.size

# Define mean squared error.
def mse(y_true, y_pred):
    errors = y_pred - y_true
    return np.mean(errors ** 2)

# Start with a flat line.
slope = 0.0
intercept = strength.mean()

# Choose learning settings.
learning_rate = 0.08
epochs = 120

# Save selected losses only.
loss_history = []
checkpoints = {0, 1, 5, 20, 119}

# Repeatedly update slope and intercept.
for epoch in range(epochs):
    predictions = slope * x + intercept
    residuals = predictions - strength

    slope_grad = (2 / n) * np.sum(residuals * x)
    intercept_grad = (2 / n) * np.sum(residuals)

    slope = slope - learning_rate * slope_grad
    intercept = intercept - learning_rate * intercept_grad

    if epoch in checkpoints:
        loss_history.append((epoch + 1, mse(strength, predictions)))

# Convert standardized slope to original units.
original_slope = slope / x_std
original_intercept = intercept - original_slope * x_mean

# Compute final predictions and residuals.
final_predictions = original_slope * days + original_intercept
final_residuals = strength - final_predictions

# Print concise learning results.
print("Gradient descent for concrete strength regression")
print(f"Final MSE: {mse(strength, final_predictions):.2f}")
print(f"Slope: {original_slope:.2f} MPa per day")
print(f"Intercept: {original_intercept:.2f} MPa")
print(f"Residual at 28 days: {final_residuals[4]:.2f} MPa")
print("Loss checkpoints:", [(e, round(v, 2)) for e, v in loss_history])

# Plot data and learned regression line.
plt.figure(figsize=(7, 4))
plt.scatter(days, strength, label="Measured strengths")
plt.plot(days, final_predictions, color="red", label="Learned line")
plt.xlabel("Curing time (days)")
plt.ylabel("Compressive strength (MPa)")
plt.title("Gradient Descent Fits a One-Feature Regression Line")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Learning Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1783738890" width="250">



>* Controls step size during gradient descent
>* Chosen setting affects training speed and stability

>* Too small learns slowly with many updates
>* Too large overshoots and destabilizes loss

>* Balance learning speed with training stability
>* Track loss to adjust step size



```
#@title Python Code - Learning Rate

# Learning rate controls gradient descent step size.
# This example uses pavement deflection data.
# We compare slow, stable, and unstable updates.

import numpy as np
import matplotlib.pyplot as plt

# Store axle load values in kips.
loads = np.array([4, 6, 8, 10, 12], dtype=float)

# Store measured deflection values in inches.
deflections = np.array([0.09, 0.13, 0.18, 0.21, 0.26], dtype=float)

# Validate matching one dimensional data shapes.
assert loads.ndim == deflections.ndim == 1
assert loads.size == deflections.size

# Scale features for steadier gradient descent.
x_values = (loads - loads.mean()) / loads.std()
y_values = deflections.copy()

# Define mean squared error for predictions.
def mse_loss(slope, intercept):
    predictions = slope * x_values + intercept
    residuals = predictions - y_values

    return np.mean(residuals ** 2)

# Run gradient descent for one learning rate.
def train_model(rate, steps=25):
    slope, intercept = 0.0, 0.0
    losses = []

    for step in range(steps):
        predictions = slope * x_values + intercept
        residuals = predictions - y_values

        slope_grad = 2 * np.mean(residuals * x_values)
        intercept_grad = 2 * np.mean(residuals)

        slope = slope - rate * slope_grad
        intercept = intercept - rate * intercept_grad

        losses.append(mse_loss(slope, intercept))
    return slope, intercept, losses

# Compare three learning rate choices.
rates = [0.01, 0.2, 1.1]
results = {rate: train_model(rate) for rate in rates}

# Print a compact summary table.
print("Learning rate comparison for pavement deflection regression")
print("rate | final slope | final intercept | final MSE")

# Show only one line per learning rate.
for rate, (slope, intercept, losses) in results.items():
    print(f"{rate:>4} | {slope:>11.4f} | {intercept:>15.4f} | {losses[-1]:.6f}")

# Plot loss curves to reveal training behavior.
plt.figure(figsize=(7, 4))
for rate, (_, _, losses) in results.items():
    plt.plot(losses, marker="o", label=f"rate={rate}")

# Label the plot for interpretation.
plt.xlabel("Gradient descent step")
plt.ylabel("Mean squared error")
plt.title("Learning rate changes descent speed and stability")
plt.legend()

# Display the single teaching plot.
plt.grid(True, alpha=0.3)
plt.show()

```

## **3. Engineering Meaning**

### **3.1. Slope Meaning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1783738882" width="250">



>* Slope shows predicted change per input unit
>* Sign indicates increase or decrease direction

>* Slope units reveal physical meaning
>* Check scale, range, and context

>* Slope shows average trends, not guarantees
>* Use with engineering judgment and uncertainty



```
#@title Python Code - Slope Meaning

# This example interprets slope in engineering.
# Road length predicts pavement repair cost.
# Units make the slope physically meaningful.

import numpy as np
import matplotlib.pyplot as plt

# Store road lengths in kilometers.
length_km = np.array([1, 2, 3, 4, 5], dtype=float)

# Store repair costs in thousand dollars.
cost_kusd = np.array([42, 57, 73, 88, 104], dtype=float)

# Validate matching one dimensional data.
assert length_km.ndim == cost_kusd.ndim == 1
assert length_km.size == cost_kusd.size

# Fit a straight line manually.
slope, intercept = np.polyfit(length_km, cost_kusd, 1)

# Compute predictions and residuals.
predicted = slope * length_km + intercept
residuals = cost_kusd - predicted

# Print concise engineering interpretation.
print(f"Slope: {slope:.2f} thousand dollars per kilometer")
print(f"Intercept: {intercept:.2f} thousand dollars at zero kilometers")
print(f"Prediction for 6 km: {slope * 6 + intercept:.2f} thousand dollars")
print(f"Residual at 3 km: {residuals[2]:.2f} thousand dollars")

# Create one plot showing slope meaning.
plt.figure(figsize=(7, 4))
plt.scatter(length_km, cost_kusd, label="Observed projects")
plt.plot(length_km, predicted, color="red", label="Linear trend")
plt.xlabel("Road length (km)")
plt.ylabel("Repair cost (thousand dollars)")
plt.title("Slope means cost increase per additional kilometer")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.2. Residual Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1783738880" width="250">



>* Residuals show prediction error size
>* They reveal where models fit poorly

>* Residual sign shows underprediction or overprediction
>* Unexpected deflections may require engineering investigation

>* Residual patterns test linear model fit
>* Patterns reveal missing factors and guide decisions



```
#@title Python Code - Residual Interpretation

# Residuals reveal model errors.
# Signs show underprediction or overprediction.
# Engineers investigate large residuals.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Store curing time observations.
days = np.array([3, 7, 14, 21, 28, 35], dtype=float)

# Store measured concrete strengths.
actual = np.array([14, 22, 31, 36, 43, 45], dtype=float)

# Define a simple regression line.
slope = 1.05
intercept = 13.0

# Compute predicted strengths.
predicted = intercept + slope * days

# Compute residuals as actual minus predicted.
residuals = actual - predicted

# Check matching data lengths.
assert days.shape == actual.shape == predicted.shape

# Find the largest absolute residual.
largest_index = int(np.argmax(np.abs(residuals)))

# Print compact engineering interpretation.
print("Residual = actual strength - predicted strength")
print(f"Slope: {slope:.2f} MPa per curing day")
print(f"Intercept: {intercept:.1f} MPa at day zero")
print(f"Largest residual occurs at day {days[largest_index]:.0f}")
print(f"Residual value: {residuals[largest_index]:+.2f} MPa")
print("Positive residual means the model underestimated strength")
print("Negative residual means the model overestimated strength")

# Create one residual plot.
plt.figure(figsize=(7, 4))
plt.axhline(0, color="black", linewidth=1)
plt.scatter(days, residuals, color="steelblue", s=70)
plt.plot(days, residuals, color="steelblue", alpha=0.5)
plt.xlabel("Curing time, days")
plt.ylabel("Residual, MPa")
plt.title("Concrete strength residuals from a linear model")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Prediction Residuals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1783738884" width="250">



>* Residuals show prediction error versus reality
>* They reveal missed engineering behavior

>* Residuals reveal missing factors and model limits
>* Patterns show bias or poor fit

>* Judge residuals by engineering context
>* Connect errors to safety and decisions



```
#@title Python Code - Prediction Residuals

# Residuals compare predictions with field measurements.
# Engineering units make errors easier to judge.
# This example uses pavement deflection data.

import numpy as np
import matplotlib.pyplot as plt

# Store axle load values in kips.
loads_kips = np.array([8, 10, 12, 14, 16, 18], dtype=float)

# Store measured deflections in inches.
measured_inches = np.array([0.030, 0.036, 0.041, 0.052, 0.057, 0.066])

# Define a simple engineering regression line.
slope = 0.0035
intercept = 0.0010

# Compute predicted pavement deflections.
predicted_inches = intercept + slope * loads_kips

# Check matching data lengths before subtraction.
assert loads_kips.shape == measured_inches.shape == predicted_inches.shape

# Residual equals observed value minus predicted value.
residuals_inches = measured_inches - predicted_inches

# Compute mean squared error in squared inches.
mse_inches = np.mean(residuals_inches ** 2)

# Find the largest absolute residual.
worst_index = int(np.argmax(np.abs(residuals_inches)))

# Print compact engineering interpretation lines.
print(f"Slope: {slope:.4f} inches per kip")
print(f"Intercept: {intercept:.4f} inches at zero load")
print(f"MSE: {mse_inches:.8f} square inches")
print(f"Worst residual: {residuals_inches[worst_index]:+.4f} inches")
print(f"At load: {loads_kips[worst_index]:.0f} kips")

# Plot measured values and model predictions.
plt.figure(figsize=(7, 4))
plt.scatter(loads_kips, measured_inches, label="Measured field data")
plt.plot(loads_kips, predicted_inches, label="Linear prediction")

# Draw residuals as vertical error segments.
for x_value, y_meas, y_pred in zip(loads_kips, measured_inches, predicted_inches):
    plt.plot([x_value, x_value], [y_pred, y_meas], color="red")

# Label the engineering plot clearly.
plt.xlabel("Axle load, kips")
plt.ylabel("Pavement deflection, inches")
plt.title("Prediction Residuals for Pavement Deflection")
plt.legend()

# Display the single teaching plot.
plt.show()

```
