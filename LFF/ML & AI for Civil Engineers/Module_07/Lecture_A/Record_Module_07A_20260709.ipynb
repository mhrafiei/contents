## **1. Network Components**

### **1.1. Neurons**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_01.jpg?v=1783622572" width="250">



>* Neurons combine inputs into one useful signal
>* Connected neurons learn richer data patterns

>* Weights control each input's influence
>* Biases set activation thresholds during training

>* Activations let neurons model nonlinear patterns
>* Many feature-sensitive neurons build complex predictions



```
#@title Python Code - Neurons

# Demonstrate one neuron using house features.
# Weights act like adjustable engineering knobs.
# Bias shifts the neuron's baseline response.

import numpy as np

# Store one house as scaled input signals.
features = np.array([0.80, 0.60, 0.20])

# Name each signal for readable output.
names = ["area", "bedrooms", "distance"]

# Choose weights for this demonstration neuron.
weights = np.array([1.20, 0.70, -0.90])

# Set a bias like an adjustable threshold.
bias = -0.30

# Validate matching input and weight shapes.
assert features.shape == weights.shape

# Compute each weighted input contribution.
contributions = features * weights

# Combine contributions and bias before activation.
raw_score = np.sum(contributions) + bias

# Apply sigmoid activation for bounded output.
activation = 1.0 / (1.0 + np.exp(-raw_score))

# Print a compact neuron summary.
print("Neuron example: house value signal")

# Show each input contribution clearly.
for name, value in zip(names, contributions):
    print(f"{name} contribution: {value:.2f}")

# Show the bias and combined score.
print(f"bias contribution: {bias:.2f}")

# Show the final activated neuron output.
print(f"raw score before activation: {raw_score:.2f}")

# Interpret the output in plain language.
message = "stronger" if activation >= 0.50 else "weaker"

# Print the final engineering interpretation.
print(f"activated output: {activation:.2f}, a {message} signal")

```

### **1.2. Weights**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_02.jpg?v=1783622574" width="250">



>* Weights control each input’s influence
>* Positive, negative, or near-zero effects

>* Weights are learned from training data
>* They tune inputs into useful predictions

>* Many weights create flexible layered pattern learning
>* Usefulness depends on data, design, and training



```
#@title Python Code - Weights

# Weights act like adjustable engineering gain settings.
# This example uses one simple artificial neuron.
# NumPy keeps the weighted calculation transparent.

import numpy as np

# Define three civil engineering input measurements.
feature_names = np.array(["load", "crack", "age"])

# Store one bridge inspection as scaled features.
inputs = np.array([0.80, 0.30, 0.60])

# Assign weights showing each feature influence.
weights = np.array([1.50, 2.00, -0.70])

# Validate matching vector lengths before multiplying.
assert inputs.shape == weights.shape

# Compute each weighted contribution separately.
contributions = inputs * weights

# Add a bias as baseline tendency.
bias = -0.40

# Sum contributions and bias before activation.
raw_score = contributions.sum() + bias

# Use sigmoid activation for probability-like output.
probability = 1.0 / (1.0 + np.exp(-raw_score))

# Print compact teaching results only.
print("Feature names:", feature_names.tolist())
print("Inputs:", inputs.round(2).tolist())
print("Weights:", weights.round(2).tolist())
print("Weighted contributions:", contributions.round(2).tolist())
print("Bias:", round(float(bias), 2))
print("Raw neuron score:", round(float(raw_score), 3))
print("Activated output:", round(float(probability), 3))
print("Largest influence:", feature_names[np.argmax(np.abs(contributions))])

```

### **1.3. Hidden Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_03.jpg?v=1783622576" width="250">



>* Hidden layers transform inputs between network stages
>* They learn useful internal feature combinations

>* Hidden layers build useful features from inputs
>* Deeper layers refine patterns step by step

>* Hidden layers model complex input relationships
>* Balance size to learn without overfitting



```
#@title Python Code - Hidden Layers

# Hidden layers transform engineering measurements.
# Each neuron combines weighted signals.
# Activations add flexible curved behavior.

# NumPy is available in Google Colab.
import numpy as np

# Use deterministic values for repeatable teaching.
np.set_printoptions(precision=3, suppress=True)

# Inputs describe one small building site.
features = np.array([[120.0, 3.0, 18.0]])

# Scale inputs into comparable engineering ranges.
scaled = np.array([[1.20, 0.60, 0.36]])

# Check the input shape before multiplying.
assert scaled.shape == (1, 3)

# Weights connect three inputs to four hidden neurons.
W_hidden = np.array([
    [0.8, -0.4, 0.3, 0.6],
    [0.2, 0.9, -0.5, 0.1],
    [-0.7, 0.3, 0.8, -0.2]
])

# Biases shift each hidden neuron threshold.
b_hidden = np.array([[0.10, -0.20, 0.05, 0.00]])

# ReLU keeps positive signals and clips negatives.
def relu(values):
    return np.maximum(values, 0.0)

# Hidden layer creates four internal features.
hidden_raw = scaled @ W_hidden + b_hidden
hidden_out = relu(hidden_raw)

# Output layer combines hidden features into prediction.
W_output = np.array([[0.7], [0.4], [-0.6], [0.5]])
b_output = np.array([[0.15]])

# Check hidden shape before final prediction.
assert hidden_out.shape == (1, 4)

# Prediction is a simple forward pass.
prediction = hidden_out @ W_output + b_output

# Compare with a direct linear model.
linear_weights = np.array([[0.45], [0.25], [-0.30]])
linear_bias = np.array([[0.10]])
linear_prediction = scaled @ linear_weights + linear_bias

# Print concise teaching results only.
print("Inputs: 120 m2, 3 bedrooms, 18 years old")
print("Hidden raw signals:", hidden_raw.round(3))
print("Hidden activated signals:", hidden_out.round(3))
print("Neural network prediction:", prediction.round(3))
print("Direct linear prediction:", linear_prediction.round(3))
print("Hidden layers create intermediate learned features.")

```

## **2. Forward Pass**

### **2.1. Matrix Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_01.jpg?v=1783622582" width="250">



>* Rows are examples; columns are features
>* Batches move through transformations toward predictions

>* Match input columns to first-layer connections
>* Preprocess data for efficient NumPy batching

>* Batches process many examples in parallel
>* Matrix inputs scale toward real predictions



```
#@title Python Code - Matrix Inputs

# Matrix inputs organize many engineering observations.
# Each row represents one apartment example.
# Columns hold consistent numerical feature measurements.

import numpy as np

# Create a small batch of apartment features.
feature_names = ["area_sqft", "bedrooms", "distance_miles"]

# Rows are apartments, columns are measured features.
X = np.array([
    [650.0, 1.0, 4.2],
    [900.0, 2.0, 2.5],
    [1200.0, 3.0, 6.0],
])

# Define first layer weights for two neurons.
W1 = np.array([
    [0.0020, -0.0010],
    [0.3000, 0.2000],
    [-0.0800, -0.0500],
])

# Biases shift each neuron before activation.
b1 = np.array([0.10, -0.20])

# Validate matrix dimensions before multiplication.
assert X.shape[1] == W1.shape[0]

# Compute all neuron inputs for the batch.
Z1 = X @ W1 + b1

# ReLU keeps positive signals and clips negatives.
A1 = np.maximum(0.0, Z1)

# Output layer combines hidden neuron activations.
W2 = np.array([[900.0], [500.0]])

# Output bias represents a base rent level.
b2 = np.array([700.0])

# Validate hidden layer dimensions before output.
assert A1.shape[1] == W2.shape[0]

# Compute predicted rent for every apartment row.
predicted_rent = A1 @ W2 + b2

# Print compact teaching results only.
print("Input matrix shape:", X.shape)
print("First layer output shape:", A1.shape)
print("Prediction shape:", predicted_rent.shape)
print("Feature order:", feature_names)
print("Predicted rents:", np.round(predicted_rent.ravel(), 2))

```

### **2.2. Array Activations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_02.jpg?v=1783622580" width="250">



>* Activations are learned signals between layers
>* Each activation array feeds the next layer

>* Activations add flexibility beyond linear sums
>* They transform raw outputs into useful signals

>* Batch activations process many examples efficiently
>* Consistent array shapes keep predictions organized



```
#@title Python Code - Array Activations

# Demonstrate activations using small engineering arrays.
# Each layer transforms measurements into signals.
# NumPy handles many examples together efficiently.

# NumPy is available in Google Colab by default.
import numpy as np

# Keep printed numbers readable for beginners.
np.set_printoptions(precision=3, suppress=True)

# Create three building examples with scaled features.
features = np.array([
    [1.20, 0.30, 0.80],
    [0.75, 0.60, 0.40],
    [1.50, 0.20, 0.15],
])

# Validate the input batch shape before multiplying.
expected_features = 3
assert features.shape[1] == expected_features

# Define first layer weights and biases.
weights_1 = np.array([
    [0.80, -0.40, 0.30, 0.10],
    [0.20, 0.90, -0.50, 0.40],
    [-0.30, 0.10, 0.70, -0.60],
])

# Biases shift each hidden neuron response.
bias_1 = np.array([0.10, -0.20, 0.05, 0.30])

# ReLU keeps positive signals and suppresses negatives.
def relu(values):
    return np.maximum(values, 0.0)

# Compute hidden layer raw scores and activations.
hidden_scores = features @ weights_1 + bias_1
hidden_activations = relu(hidden_scores)

# Define output layer weights and bias.
weights_2 = np.array([[0.60], [-0.20], [0.50], [0.30]])
bias_2 = np.array([0.15])

# Convert hidden activations into final predictions.
output_scores = hidden_activations @ weights_2 + bias_2
predictions = output_scores.reshape(-1)

# Print compact shape information and results.
print("Input shape:", features.shape)
print("Hidden activation shape:", hidden_activations.shape)
print("Output prediction shape:", predictions.shape)
print("Hidden activations:\n", hidden_activations)
print("Predicted relative demand:", predictions)

```

### **2.3. Output Prediction**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_03.jpg?v=1783622584" width="250">



>* Final layer turns computations into predictions
>* Output shape and meaning depend on task

>* Track output shapes for single or batched inputs
>* Match prediction arrays to original examples

>* Interpret outputs by task and activation
>* Use thresholds and context for decisions



```
#@title Python Code - Output Prediction

# This example predicts bridge maintenance risk.
# NumPy arrays carry batches through layers.
# Output values become engineering decisions.

# NumPy is already available in Colab.
import numpy as np

# Store three bridge inspection examples.
X = np.array([
    [0.20, 0.10, 0.30],
    [0.70, 0.60, 0.80],
    [0.40, 0.90, 0.50]
])

# Validate the input batch shape.
assert X.shape == (3, 3)

# Define first layer weights and biases.
W1 = np.array([
    [0.8, -0.4, 0.3, 0.2],
    [0.5, 0.7, -0.6, 0.1],
    [-0.3, 0.2, 0.9, 0.4]
])

# Define hidden layer bias values.
b1 = np.array([0.1, -0.2, 0.05, 0.0])

# Define output layer weights and bias.
W2 = np.array([[1.2], [-0.7], [0.9], [0.4]])
b2 = np.array([-0.1])

# Validate compatible matrix dimensions.
assert X.shape[1] == W1.shape[0]
assert W1.shape[1] == W2.shape[0]

# Apply ReLU activation to hidden signals.
hidden_raw = X @ W1 + b1
hidden = np.maximum(0, hidden_raw)

# Convert final signal into probability.
output_raw = hidden @ W2 + b2
risk = 1 / (1 + np.exp(-output_raw))

# Convert probabilities into simple decisions.
labels = np.where(risk >= 0.5, "inspect", "monitor")

# Print compact prediction results.
print("Input shape:", X.shape)
print("Output shape:", risk.shape)
print("Risk scores:", np.round(risk.ravel(), 3))
print("Decisions:", labels.ravel().tolist())
print("Each output row matches one input bridge.")

```

## **3. Model Flexibility**

### **3.1. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_01.jpg?v=1783622568" width="250">



>* Linear models use straight decision boundaries
>* Neural networks capture curved, complex patterns

>* Hidden layers build useful intermediate patterns
>* Networks capture complex boundaries without manual features

>* Flexible networks need careful overfitting control
>* Best for complex interacting data patterns



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Neural networks combine simple nonlinear neuron responses.
# This example uses synthetic engineering inspection data.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic seed for repeatable results.
rng = np.random.default_rng(7)

# Create a small grid of inspection locations.
x_values = np.linspace(-3.0, 3.0, 160)
y_values = np.linspace(-3.0, 3.0, 160)

# Build coordinate matrices for plotting boundaries.
xx, yy = np.meshgrid(x_values, y_values)
points = np.column_stack([xx.ravel(), yy.ravel()])

# Validate the grid before model calculations.
assert points.shape[1] == 2 and points.shape[0] > 100

# Define a simple linear classifier score.
linear_score = points[:, 0] + points[:, 1] - 0.2
linear_class = (linear_score > 0).reshape(xx.shape)

# Define sigmoid activation for hidden neurons.
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# Choose fixed weights for a tiny neural network.
W1 = np.array([[3.0, 0.0], [-3.0, 0.0], [0.0, 3.0], [0.0, -3.0]])
b1 = np.array([-2.0, -2.0, -2.0, -2.0])

# Combine hidden responses into a nonlinear score.
W2 = np.array([1.0, 1.0, 1.0, 1.0])
b2 = -1.35

# Run forward prediction through hidden neurons.
hidden = sigmoid(points @ W1.T + b1)
network_score = hidden @ W2 + b2
network_class = (network_score > 0).reshape(xx.shape)

# Create sample inspection points around two regions.
angles = rng.uniform(0.0, 2.0 * np.pi, 80)
radii = rng.normal(1.8, 0.18, 80)

# Convert polar samples into two feature columns.
sample_x = radii * np.cos(angles)
sample_y = radii * np.sin(angles)
samples = np.column_stack([sample_x, sample_y])

# Label points using a ring shaped rule.
distance = np.sqrt(sample_x**2 + sample_y**2)
labels = (distance > 1.55).astype(int)

# Print a compact teaching summary.
print("Linear boundary: one straight separating rule.")
print("Neural network boundary: combined curved regions.")
print(f"Grid points evaluated: {points.shape[0]}")
print(f"Hidden neurons used: {hidden.shape[1]}")

# Plot both decision boundaries side by side.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Show the linear model boundary.
axes[0].contourf(xx, yy, linear_class, levels=1, alpha=0.35)
axes[0].scatter(sample_x, sample_y, c=labels, s=18, edgecolor="k")
axes[0].set_title("Linear model")
axes[0].set_xlabel("Sensor offset, inches")

# Show the neural network boundary.
axes[1].contourf(xx, yy, network_class, levels=1, alpha=0.35)
axes[1].scatter(sample_x, sample_y, c=labels, s=18, edgecolor="k")
axes[1].set_title("Tiny neural network")
axes[1].set_xlabel("Sensor offset, inches")

# Keep axes comparable for visual interpretation.
for axis in axes:
    axis.set_ylabel("Vibration change, mm/s")
    axis.set_aspect("equal", adjustable="box")
    axis.grid(True, alpha=0.25)

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```

### **3.2. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_02.jpg?v=1783622566" width="250">



>* Linear models split data with straight boundaries
>* Real patterns may need curved boundaries

>* Neural networks create flexible curved boundaries
>* They learn complex feature interactions automatically

>* Balance underfitting and overfitting risks
>* Choose flexibility only when it improves generalization



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Tiny synthetic data keeps runtime predictable.
# NumPy builds models without ML libraries.

import numpy as np
import matplotlib.pyplot as plt

# Set a seed for repeatable points.
rng = np.random.default_rng(7)

# Create two circular engineering risk zones.
n_points = 160
angles = rng.uniform(0.0, 2.0 * np.pi, n_points)

radii = rng.normal(1.0, 0.12, n_points)
inner_x = np.column_stack((radii * np.cos(angles), radii * np.sin(angles)))

outer_radii = rng.normal(2.0, 0.16, n_points)
outer_x = np.column_stack((outer_radii * np.cos(angles), outer_radii * np.sin(angles)))

# Stack features and labels together.
X = np.vstack((inner_x, outer_x))
y = np.r_[np.zeros(n_points), np.ones(n_points)]

# Validate the small teaching dataset.
assert X.shape == (2 * n_points, 2)
assert y.shape == (2 * n_points,)

# Linear score gives one straight boundary.
linear_score = X[:, 0] + 0.15 * X[:, 1] - 0.2
linear_pred = (linear_score > 0.0).astype(int)

# Nonlinear score uses distance from origin.
nonlinear_score = X[:, 0] ** 2 + X[:, 1] ** 2 - 2.25
nonlinear_pred = (nonlinear_score > 0.0).astype(int)

# Compute simple accuracy values.
linear_acc = np.mean(linear_pred == y)
nonlinear_acc = np.mean(nonlinear_pred == y)

print(f"Dataset shape: {X.shape[0]} points, {X.shape[1]} features")
print(f"Straight boundary accuracy: {linear_acc:.2f}")
print(f"Curved boundary accuracy: {nonlinear_acc:.2f}")
print("Key idea: nonlinear boundaries can wrap around clusters.")

# Build a compact plotting grid.
grid = np.linspace(-2.7, 2.7, 220)
xx, yy = np.meshgrid(grid, grid)

# Evaluate both boundaries on the grid.
linear_grid = xx + 0.15 * yy - 0.2
nonlinear_grid = xx ** 2 + yy ** 2 - 2.25

# Draw one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the straight boundary panel.
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", s=18)
axes[0].contour(xx, yy, linear_grid, levels=[0], colors="black")
axes[0].set_title("Linear model: straight split")
axes[0].set_aspect("equal")

# Plot the curved boundary panel.
axes[1].scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", s=18)
axes[1].contour(xx, yy, nonlinear_grid, levels=[0], colors="black")
axes[1].set_title("Neural-network-like: curved split")
axes[1].set_aspect("equal")

# Label axes as two sensor measurements.
for axis in axes:
    axis.set_xlabel("Sensor A reading")
    axis.set_ylabel("Sensor B reading")

plt.tight_layout()
plt.show()

```

### **3.3. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_03.jpg?v=1783622570" width="250">



>* Linear models use rigid decision boundaries
>* Neural networks learn flexible curved boundaries

>* Hidden layers learn complex feature patterns
>* Networks form curved, irregular class boundaries

>* Simple models are clearer and safer
>* Neural networks need careful validation



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Neural networks can bend classification regions.
# This example uses only NumPy.

import numpy as np
import matplotlib.pyplot as plt

# Create a small two dimensional grid.
x_values = np.linspace(-3.0, 3.0, 220)
y_values = np.linspace(-3.0, 3.0, 220)
xx, yy = np.meshgrid(x_values, y_values)

# Stack coordinates as model input features.
points = np.column_stack((xx.ravel(), yy.ravel()))
assert points.shape[1] == 2

# Define a simple linear model boundary.
linear_score = points[:, 0] + points[:, 1] - 0.2
linear_class = linear_score > 0.0

# Define fixed hidden neurons manually.
centers = np.array([[1.2, 1.0], [-1.1, -0.9]])
width = 0.9

# Hidden activations respond near two clusters.
diff = points[:, None, :] - centers[None, :, :]
squared_distance = np.sum(diff * diff, axis=2)
hidden = np.exp(-squared_distance / (2.0 * width * width))

# Combine hidden responses into nonlinear predictions.
network_score = hidden[:, 0] + hidden[:, 1] - 0.55
network_class = network_score > 0.0

# Reshape predictions back onto the grid.
linear_grid = linear_class.reshape(xx.shape)
network_grid = network_class.reshape(xx.shape)

# Print a compact engineering interpretation.
print("Linear model: one straight separating cut.")
print("Neural network: two curved high-risk pockets.")
print("Hidden neurons act like pattern detectors.")

# Plot both decision regions side by side.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Show the rigid linear decision boundary.
axes[0].contourf(xx, yy, linear_grid, levels=1, alpha=0.35)
axes[0].contour(xx, yy, linear_score.reshape(xx.shape), levels=[0])
axes[0].set_title("Linear boundary")
axes[0].set_xlabel("Feature 1")

# Show the flexible neural network boundary.
axes[1].contourf(xx, yy, network_grid, levels=1, alpha=0.35)
axes[1].contour(xx, yy, network_score.reshape(xx.shape), levels=[0])
axes[1].set_title("Nonlinear neural boundary")
axes[1].set_xlabel("Feature 1")

# Keep axes comparable for visual comparison.
for axis in axes:
    axis.set_ylabel("Feature 2")
    axis.set_aspect("equal")
    axis.grid(True, alpha=0.25)

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```
