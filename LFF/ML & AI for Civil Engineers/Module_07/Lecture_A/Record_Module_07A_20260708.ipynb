## **1. Network Components**

### **1.1. Neurons**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_01.jpg?v=1783565385" width="250">



>* Neuron combines inputs into one output
>* Learned settings transform data for predictions

>* Neurons learn which inputs matter most
>* Training tunes responses using data

>* Neurons combine to detect complex patterns
>* Simple math units transform data into outputs



```
#@title Python Code - Neurons

# A neuron combines engineering measurements.
# Weights control each input influence.
# Bias shifts the decision threshold.

import numpy as np

# Store three normalized house features.
inputs = np.array([0.80, 0.60, 0.30])

# Store learned influence settings.
weights = np.array([1.20, 0.70, -0.50])

# Store one adjustable offset value.
bias = -0.40

# Validate matching vector lengths.
assert inputs.shape == weights.shape

# Combine inputs using weighted addition.
weighted_sum = np.dot(inputs, weights) + bias

# Convert combined signal through activation.
neuron_output = 1 / (1 + np.exp(-weighted_sum))

# Print compact teaching results.
print("Inputs: location, area, age")
print("Weights show learned importance")
print(f"Weighted sum plus bias: {weighted_sum:.3f}")
print(f"Neuron output signal: {neuron_output:.3f}")
print("Higher output means stronger positive evidence")

```

### **1.2. Weights**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_02.jpg?v=1783565387" width="250">



>* Weights control each input’s influence
>* They mix signals for the next layer

>* Training adjusts weights using prediction errors
>* Weights store learned relationships numerically

>* Many weights build complex pattern representations
>* Distributed decisions increase power but reduce interpretability



```
#@title Python Code - Weights

# Weights act like adjustable engineering gain knobs.
# This example mixes three civil engineering signals.
# Larger weights create stronger neuron influence.

import numpy as np
import matplotlib.pyplot as plt

# Store one inspection case as input features.
features = np.array([0.80, 0.30, 0.60])
feature_names = ["crack width", "traffic load", "drainage risk"]

# Validate the feature vector before calculations.
assert features.shape == (3,), "Expected exactly three input features."
assert len(feature_names) == features.size, "Names must match features."

# Define three different weight settings.
weight_sets = np.array([
    [0.20, 0.20, 0.20],
    [1.20, 0.20, 0.20],
    [0.20, 1.20, -0.80],
])

# Add one bias shared by each neuron example.
bias = -0.25
labels = ["balanced", "crack emphasized", "traffic versus drainage"]

# Compute weighted sums for each setting.
weighted_sums = weight_sets @ features + bias
activations = 1 / (1 + np.exp(-weighted_sums))

# Print a compact teaching summary.
print("Inputs: crack=0.80, traffic=0.30, drainage=0.60")
print("Bias shifts every weighted sum by -0.25")

# Show how weights change neuron output.
for label, total, output in zip(labels, weighted_sums, activations):
    print(f"{label}: sum={total:.2f}, activation={output:.2f}")

# Plot the three activation results.
plt.figure(figsize=(7, 4))
plt.bar(labels, activations, color=["gray", "steelblue", "orange"])
plt.ylim(0, 1)
plt.ylabel("Neuron activation")
plt.title("Changing weights changes the same neuron's response")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()

```

### **1.3. Hidden Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_01_03.jpg?v=1783565389" width="250">



>* Hidden layers transform inputs between network stages
>* They learn useful patterns from sensor data

>* Hidden layers build features step by step
>* Training discovers useful internal patterns

>* Hidden layers capture complex input relationships
>* Larger networks need data and care



```
#@title Python Code - Hidden Layers

# Hidden layers transform raw engineering measurements.
# Each neuron mixes inputs differently.
# Activations create flexible internal features.

import numpy as np
import matplotlib.pyplot as plt

# Store two sensor readings for one machine.
inputs = np.array([82.0, 0.42])
assert inputs.shape == (2,)

# Scale temperature and vibration for stable arithmetic.
scaled_inputs = np.array([inputs[0] / 100.0, inputs[1]])
assert scaled_inputs.shape == (2,)

# Define three hidden neurons with different weights.
hidden_weights = np.array([[1.2, 2.0], [-0.8, 1.5], [0.5, -1.0]])
hidden_biases = np.array([-1.0, 0.2, 0.1])
assert hidden_weights.shape == (3, 2)

# Compute hidden layer signals before activation.
hidden_raw = hidden_weights @ scaled_inputs + hidden_biases
hidden_activated = np.maximum(0.0, hidden_raw)
assert hidden_activated.shape == (3,)

# Combine hidden features into one risk score.
output_weights = np.array([1.4, -0.6, 0.8])
output_bias = -0.2
risk_score = output_weights @ hidden_activated + output_bias

# Convert score into probability using sigmoid activation.
failure_probability = 1.0 / (1.0 + np.exp(-risk_score))
print(f"Inputs: temperature={inputs[0]:.0f} C, vibration={inputs[1]:.2f} g")
print(f"Hidden features: {np.round(hidden_activated, 3)}")
print(f"Predicted failure probability: {failure_probability:.2%}")

# Plot hidden neuron outputs as internal features.
labels = ["Neuron 1", "Neuron 2", "Neuron 3"]
plt.figure(figsize=(6, 3))
plt.bar(labels, hidden_activated, color="steelblue")
plt.ylabel("Activated hidden value")
plt.title("Hidden layer creates intermediate features")
plt.ylim(0, max(1.0, hidden_activated.max() + 0.2))
plt.show()

```

## **2. Forward Pass**

### **2.1. Matrix Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_01.jpg?v=1783565379" width="250">



>* Store examples as rows, features as columns
>* Use NumPy to process batches efficiently

>* Weights connect input features to neurons
>* NumPy processes batches with fast matrix operations

>* Match input features to layer weights
>* Track shapes to debug forward passes



```
#@title Python Code - Matrix Inputs

# Matrix inputs organize many examples together.
# Rows are examples, columns are features.
# Shapes guide safe neural network calculations.

# Import NumPy for compact matrix operations.
import numpy as np

# Create three apartment examples with four features.
X = np.array([
    [850, 2, 3.2, 12],
    [1200, 3, 1.1, 5],

    [650, 1, 5.0, 30]
], dtype=float)

# Scale features to comparable engineering ranges.
scale = np.array([1000, 5, 10, 50], dtype=float)
X_scaled = X / scale

# Define weights for four features and three neurons.
W1 = np.array([
    [0.8, -0.3, 0.2],
    [0.4, 0.7, -0.5],

    [-0.6, 0.2, 0.9],
    [-0.2, -0.4, 0.3]
], dtype=float)

# Define one bias value for each neuron.
b1 = np.array([0.1, -0.2, 0.05], dtype=float)

# Check that feature counts align before multiplying.
assert X_scaled.shape[1] == W1.shape[0], "Feature and weight shapes mismatch."

# Compute hidden layer values for the whole batch.
Z1 = X_scaled @ W1 + b1
A1 = np.maximum(0, Z1)

# Define output weights for one predicted rent score.
W2 = np.array([[1.2], [-0.7], [0.5]], dtype=float)
b2 = np.array([0.3], dtype=float)

# Check hidden neuron counts before final multiplication.
assert A1.shape[1] == W2.shape[0], "Hidden and output shapes mismatch."

# Compute one output for every apartment row.
y_pred = A1 @ W2 + b2

# Print compact shape information and predictions.
print("Input matrix shape:", X_scaled.shape)
print("First layer weight shape:", W1.shape)
print("Hidden output shape:", A1.shape)
print("Prediction output shape:", y_pred.shape)
print("Predicted rent scores:", np.round(y_pred.ravel(), 3))

```

### **2.2. Array Activations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_02.jpg?v=1783565381" width="250">



>* Activations transform raw neuron scores
>* Arrays process whole layers efficiently

>* Activations add flexible nonlinear behavior
>* Arrays apply activations efficiently across layers

>* Track array shapes and neuron meanings
>* Hidden activations transform data toward predictions



```
#@title Python Code - Array Activations

# Demonstrate array activations in one forward pass.
# Use tiny civil engineering style input data.
# Keep shapes visible and outputs compact.

import numpy as np

# Store three bridge cases as input rows.
X = np.array([
    [30.0, 12.0, 0.20],
    [45.0, 18.0, 0.35],
    [60.0, 25.0, 0.50],
])

# Scale inputs for stable neuron scores.
X_scaled = X / np.array([60.0, 30.0, 1.0])

# Define weights from inputs to hidden neurons.
W1 = np.array([
    [0.8, -0.4, 0.3, 0.6],
    [0.2, 0.7, -0.5, 0.1],
    [-0.6, 0.3, 0.9, -0.2],
])

# Define one bias for each hidden neuron.
b1 = np.array([0.10, -0.20, 0.05, 0.00])

# Check matrix shapes before multiplication.
assert X_scaled.shape[1] == W1.shape[0]

# Compute raw hidden layer scores.
z1 = X_scaled @ W1 + b1

# Apply ReLU activation to the whole array.
a1 = np.maximum(0.0, z1)

# Define weights from hidden layer to output.
W2 = np.array([[0.7], [-0.3], [0.5], [0.2]])

# Define output bias for final prediction.
b2 = np.array([0.15])

# Check hidden activation shape before output layer.
assert a1.shape[1] == W2.shape[0]

# Compute final forward pass predictions.
y_pred = a1 @ W2 + b2

# Print compact shape information for learners.
print("Input shape:", X_scaled.shape)
print("Raw hidden scores shape:", z1.shape)
print("Activated hidden array shape:", a1.shape)
print("Output prediction shape:", y_pred.shape)

# Print rounded arrays without overwhelming output.
print("First hidden raw scores:", np.round(z1[0], 3))
print("First hidden activations:", np.round(a1[0], 3))
print("Predicted risk scores:", np.round(y_pred.ravel(), 3))

```

### **2.3. Output Prediction**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_02_03.jpg?v=1783565383" width="250">



>* Final layer turns activations into predictions
>* Outputs are NumPy arrays with task meanings

>* Regression outputs direct numerical predictions
>* Classification outputs category scores or probabilities

>* Keep batch outputs matched to inputs
>* Check shape, range, and meaning



```
#@title Python Code - Output Prediction

# Demonstrate neural network output prediction.
# Use tiny civil engineering style data.
# Keep array shapes visible and simple.

import numpy as np

# Store three bridge inspection examples.
X = np.array([
    [12.0, 0.30, 45.0],
    [25.0, 0.55, 70.0],

    [8.0, 0.20, 30.0]
])

# Confirm rows are examples.
assert X.shape == (3, 3)

# Define first layer weights.
W1 = np.array([
    [0.04, -0.03, 0.02, 0.01],
    [1.20, 0.80, -0.50, 0.30],

    [0.01, 0.02, 0.03, -0.02]
])

# Define first layer biases.
b1 = np.array([0.10, -0.20, 0.05, 0.00])

# Define output layer weights.
W2 = np.array([
    [0.60],
    [-0.40],

    [0.30],
    [0.20]
])

# Define output layer bias.
b2 = np.array([0.15])

# Check matrix dimensions before multiplying.
assert X.shape[1] == W1.shape[0]
assert W1.shape[1] == W2.shape[0]

# Compute hidden layer preactivation values.
z1 = X @ W1 + b1

# Apply ReLU activation elementwise.
hidden = np.maximum(0, z1)

# Compute final network output predictions.
predicted_risk = hidden @ W2 + b2

# Keep one prediction for each bridge.
assert predicted_risk.shape == (3, 1)

# Print compact teaching results.
print("Input shape:", X.shape)
print("Hidden activation shape:", hidden.shape)
print("Output prediction shape:", predicted_risk.shape)
print("Predicted risk scores:", predicted_risk.ravel().round(2))
print("Highest risk bridge index:", int(np.argmax(predicted_risk)))

```

## **3. Model Flexibility**

### **3.1. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_01.jpg?v=1783565391" width="250">



>* Linear models use simple straight boundaries
>* Neural networks learn flexible curved boundaries

>* Layers combine simple patterns into complex boundaries
>* Networks capture curves, clusters, and local features

>* Simpler models can generalize more reliably
>* Neural networks capture complexity but may overfit



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Neural networks combine simple nonlinear neuron responses.
# This tiny example uses synthetic engineering points.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make results deterministic for teaching.
rng = np.random.default_rng(7)

# Create small coordinate grid in meters.
x_values = np.linspace(-3.0, 3.0, 160)
y_values = np.linspace(-3.0, 3.0, 160)

# Build two dimensional map coordinates.
xx, yy = np.meshgrid(x_values, y_values)
points = np.c_[xx.ravel(), yy.ravel()]

# Validate the grid shape before prediction.
assert points.shape[1] == 2 and points.shape[0] > 100

# Define a simple linear boundary score.
linear_score = points[:, 0] + points[:, 1] - 0.2
linear_class = (linear_score > 0).reshape(xx.shape)

# Define fixed neural network weights.
W1 = np.array([[1.4, -1.2, 0.9, -0.8], [1.1, 1.3, -1.0, -1.4]])
b1 = np.array([-1.0, -0.8, -0.7, -0.9])

# Define output weights and bias.
W2 = np.array([[1.2], [1.1], [1.0], [1.3]])
b2 = np.array([-1.15])

# Use sigmoid activation for nonlinear bending.
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# Run forward prediction through hidden layer.
hidden = sigmoid(points @ W1 + b1)
network_score = hidden @ W2 + b2

# Convert network scores into predicted classes.
network_class = (network_score.ravel() > 0).reshape(xx.shape)

# Sample a few inspection points.
sample_points = np.array([[0.0, 0.0], [2.0, 0.0], [-2.0, 1.5]])
sample_hidden = sigmoid(sample_points @ W1 + b1)

# Predict sample classes using both models.
sample_linear = sample_points[:, 0] + sample_points[:, 1] - 0.2
sample_network = sample_hidden @ W2 + b2

# Print concise teaching summary.
print("NumPy version:", np.__version__)
print("Linear boundary: one straight separation line.")
print("Neural network boundary: curved regions from hidden neurons.")
print("Sample points are measured in meters.")
print("Linear classes:", (sample_linear > 0).astype(int).tolist())
print("Network classes:", (sample_network.ravel() > 0).astype(int).tolist())

# Create exactly one comparison plot.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the linear model boundary.
axes[0].contourf(xx, yy, linear_class, levels=1, alpha=0.35)
axes[0].contour(xx, yy, linear_score.reshape(xx.shape), levels=[0], colors="black")
axes[0].set_title("Linear model")

# Plot the neural network boundary.
axes[1].contourf(xx, yy, network_class, levels=1, alpha=0.35)
axes[1].contour(xx, yy, network_score.reshape(xx.shape), levels=[0], colors="black")
axes[1].set_title("Small neural network")

# Label both engineering coordinate maps.
for axis in axes:
    axis.set_xlabel("east-west position, meters")
    axis.set_ylabel("north-south position, meters")
    axis.set_aspect("equal")

# Display the single teaching figure.
plt.tight_layout()
plt.show()

```

### **3.2. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_02.jpg?v=1783565392" width="250">



>* Linear models use straight decision boundaries
>* Real problems often need curved boundaries

>* Layers create flexible curved boundaries
>* Networks learn complex feature interactions from data

>* Balance flexibility against underfitting and overfitting
>* Choose models based on complexity and transparency



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Tiny NumPy networks show model flexibility.
# Civil examples often contain nonlinear interactions.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make the example deterministic.
rng = np.random.default_rng(7)

# Create a small two dimensional grid.
x1 = np.linspace(-3.0, 3.0, 220)
x2 = np.linspace(-3.0, 3.0, 220)

# Combine coordinates into model inputs.
X1, X2 = np.meshgrid(x1, x2)
X = np.column_stack([X1.ravel(), X2.ravel()])

# Validate the input matrix shape.
assert X.shape[1] == 2 and X.shape[0] > 100

# Define a simple linear score.
linear_score = 0.9 * X[:, 0] - 0.7 * X[:, 1]
linear_pred = linear_score > 0.0

# Define a tiny neural network manually.
W1 = np.array([[1.4, -1.2, 0.8], [1.1, 1.3, -1.5]])
b1 = np.array([-0.2, -0.1, 0.4])

# Use tanh as nonlinear activation.
hidden = np.tanh(X @ W1 + b1)
W2 = np.array([1.2, 1.1, 1.0])

# Convert hidden patterns into predictions.
nn_score = hidden @ W2 - 0.55
nn_pred = nn_score > 0.0

# Reshape predictions for contour plotting.
linear_grid = linear_pred.reshape(X1.shape)
nn_grid = nn_pred.reshape(X1.shape)

# Print a compact teaching summary.
print("Linear model boundary: one straight separation.")
print("Neural network boundary: curved regions from activations.")
print("Inputs could represent load ratio and moisture level.")

# Create one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the linear decision regions.
axes[0].contourf(X1, X2, linear_grid, levels=1, alpha=0.35)
axes[0].contour(X1, X2, linear_score.reshape(X1.shape), levels=[0])
axes[0].set_title("Linear boundary")

# Plot the neural network regions.
axes[1].contourf(X1, X2, nn_grid, levels=1, alpha=0.35)
axes[1].contour(X1, X2, nn_score.reshape(X1.shape), levels=[0])
axes[1].set_title("Neural network boundary")

# Label axes using engineering style features.
for ax in axes:
    ax.set_xlabel("Load ratio")
    ax.set_ylabel("Moisture level")
    ax.set_aspect("equal")

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```

### **3.3. Nonlinear Decision Boundaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_07/Lecture_A/image_03_03.jpg?v=1783565394" width="250">



>* Linear models create straight class boundaries
>* Real problems often need curved boundaries

>* Layers build flexible nonlinear boundaries
>* Networks capture complex feature interactions

>* Flexible boundaries need careful control
>* Use simpler models when appropriate



```
#@title Python Code - Nonlinear Decision Boundaries

# Compare straight and curved decision boundaries.
# Neural networks can bend classification regions.
# This example uses only NumPy arrays.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Make the example deterministic and small.
rng = np.random.default_rng(7)
n_points = 160

# Create two engineering-style input measurements.
x1 = rng.uniform(-2.5, 2.5, n_points)
x2 = rng.uniform(-2.5, 2.5, n_points)

# Label points inside a circular safe zone.
radius = np.sqrt(x1**2 + x2**2)
y = (radius < 1.35).astype(int)

# Stack features into one design matrix.
X = np.column_stack((x1, x2))
assert X.shape == (n_points, 2)

# Define a simple linear classifier manually.
linear_score = 0.9 * X[:, 0] - 0.4 * X[:, 1]
linear_pred = (linear_score > 0.0).astype(int)

# Define sigmoid activation for hidden neurons.
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# Build fixed hidden neurons detecting four directions.
W1 = np.array([[3.0, 0.0], [-3.0, 0.0], [0.0, 3.0], [0.0, -3.0]])
b1 = np.array([2.0, 2.0, 2.0, 2.0])

# Combine hidden detections into a curved boundary.
hidden = sigmoid(X @ W1.T + b1)
network_score = hidden.sum(axis=1) - 3.15

# Convert network scores into class predictions.
network_pred = (network_score > 0.0).astype(int)
linear_acc = np.mean(linear_pred == y)
network_acc = np.mean(network_pred == y)

# Print only a few teaching summary lines.
print(f"Data shape: {X.shape}, labels: inside versus outside circle")
print(f"Linear boundary accuracy: {linear_acc:.2f}")
print(f"Small neural network accuracy: {network_acc:.2f}")
print("The neural network bends the boundary using hidden activations.")

# Prepare a grid for visual decision regions.
grid = np.linspace(-2.7, 2.7, 180)
g1, g2 = np.meshgrid(grid, grid)
G = np.column_stack((g1.ravel(), g2.ravel()))

# Predict both models across the grid.
linear_grid = (0.9 * G[:, 0] - 0.4 * G[:, 1] > 0.0)
hidden_grid = sigmoid(G @ W1.T + b1)
network_grid = (hidden_grid.sum(axis=1) - 3.15 > 0.0)

# Draw exactly one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
models = [(linear_grid, "Linear model"), (network_grid, "Neural network")]

# Plot each decision boundary and the same data.
for ax, (pred_grid, title) in zip(axes, models):
    ax.contourf(g1, g2, pred_grid.reshape(g1.shape), alpha=0.25)
    ax.scatter(x1, x2, c=y, s=25, edgecolor="k")
    ax.set_title(title)
    ax.set_xlabel("Sensor A deviation, meters")
    ax.set_ylabel("Sensor B deviation, meters")
    ax.set_aspect("equal")

# Display the comparison plot.
plt.tight_layout()
plt.show()

```
