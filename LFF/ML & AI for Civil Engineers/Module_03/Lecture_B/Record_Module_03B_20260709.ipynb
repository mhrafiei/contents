## **1. Regression Error Metrics**

### **1.1. Regression Residuals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1783614232" width="250">



>* Residuals show each prediction’s error
>* They reveal underestimates and overestimates

>* Residuals show error size and direction
>* Patterns reveal systematic prediction mistakes

>* Residual patterns reveal model quality
>* They show where predictions are reliable



```
#@title Python Code - Regression Residuals

# Residuals compare actual and predicted values.
# Direction shows overprediction or underprediction clearly.
# Small examples help engineers inspect errors.

# Store actual bridge deflections in millimeters.
actual_mm = [12.0, 18.5, 25.0, 31.5, 40.0]

# Store predicted bridge deflections in millimeters.
predicted_mm = [11.0, 20.0, 23.5, 35.0, 38.0]

# Validate both lists before calculating residuals.
if len(actual_mm) != len(predicted_mm):
    raise ValueError("Actual and predicted lists must match.")

# Calculate residuals as actual minus predicted.
residuals_mm = []
for actual, predicted in zip(actual_mm, predicted_mm):
    residuals_mm.append(actual - predicted)

# Label each residual direction for interpretation.
directions = []
for residual in residuals_mm:
    if residual > 0:
        directions.append("underpredicted")
    elif residual < 0:
        directions.append("overpredicted")
    else:
        directions.append("exact")

# Print a compact residual table.
print("Case | Actual | Predicted | Residual | Direction")
for index, values in enumerate(zip(actual_mm, predicted_mm, residuals_mm, directions), 1):
    actual, predicted, residual, direction = values
    print(f"{index} | {actual:.1f} | {predicted:.1f} | {residual:+.1f} | {direction}")

# Summarize average signed residual.
mean_residual = sum(residuals_mm) / len(residuals_mm)
print(f"Mean residual: {mean_residual:+.2f} mm")

```

### **1.2. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1783614235" width="250">



>* MSE averages squared prediction errors
>* Lower MSE means closer predictions

>* Squaring prevents opposite errors from canceling
>* Large mistakes receive stronger penalties

>* Subtract, square, then average prediction errors
>* Lower MSE helps compare similar models



```
#@title Python Code - Mean Squared Error

# Manual metrics reveal model error.
# Squared residuals penalize large mistakes.
# Small civil examples support interpretation.

# Define observed concrete strengths.
actual_mpa = [28, 32, 35, 40, 45]

# Define predicted concrete strengths.
predicted_mpa = [30, 31, 37, 38, 44]

# Validate matching list lengths.
if len(actual_mpa) != len(predicted_mpa):
    raise ValueError("Lists must have matching lengths")

# Validate at least one observation.
if len(actual_mpa) == 0:
    raise ValueError("At least one observation is required")

# Compute residuals manually.
residuals = []
for actual, predicted in zip(actual_mpa, predicted_mpa):
    residuals.append(actual - predicted)

# Square each residual manually.
squared_errors = []
for residual in residuals:
    squared_errors.append(residual ** 2)

# Average squared errors manually.
mse_value = sum(squared_errors) / len(squared_errors)

# Show compact teaching results.
print("Actual strengths MPa:", actual_mpa)
print("Predicted strengths MPa:", predicted_mpa)
print("Residuals MPa:", residuals)
print("Squared errors:", squared_errors)
print("Manual MSE:", round(mse_value, 2), "MPa squared")
print("Lower MSE means closer regression predictions.")

```

### **1.3. Error Calculation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1783614237" width="250">



>* Compare each prediction with its actual value
>* Summarize errors to judge overall performance

>* Square errors before averaging them
>* Large mistakes affect MSE more

>* Manually compute MSE step by step
>* Interpret errors within real problem context



```
#@title Python Code - Error Calculation

# Manual errors reveal regression model quality.
# Squared errors prevent cancellation between residuals.
# Mean squared error summarizes prediction mistakes.

# Store actual concrete strengths in megapascals.
actual_strengths = [28.0, 32.5, 35.0, 40.0, 45.5]

# Store predicted strengths from a simple model.
predicted_strengths = [30.0, 31.0, 36.5, 38.0, 48.0]

# Validate matching list lengths before calculation.
if len(actual_strengths) != len(predicted_strengths):
    raise ValueError("Actual and predicted lists must match.")

# Validate that at least one prediction exists.
if len(actual_strengths) == 0:
    raise ValueError("At least one prediction is required.")

# Calculate residuals for each paired observation.
errors = []
for actual, predicted in zip(actual_strengths, predicted_strengths):
    errors.append(predicted - actual)

# Square each residual to remove direction.
squared_errors = []
for error in errors:
    squared_errors.append(error ** 2)

# Average squared errors manually for MSE.
total_squared_error = sum(squared_errors)
number_of_predictions = len(squared_errors)
mean_squared_error = total_squared_error / number_of_predictions

# Print a compact teaching summary.
print("Actual strengths:", actual_strengths)
print("Predicted strengths:", predicted_strengths)
print("Residual errors:", errors)
print("Squared errors:", squared_errors)
print("Manual MSE:", round(mean_squared_error, 2))
print("Lower MSE means predictions are closer overall.")

```

## **2. Classification Metrics**

### **2.1. Confusion Matrix**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1783614245" width="250">



>* Compares predictions with true class labels
>* Shows true and false classification outcomes

>* Error types have different real-world consequences
>* Matrix counts guide metric interpretation and adjustments

>* Works for multiple classification categories
>* Reveals hidden errors beyond overall accuracy



```
#@title Python Code - Confusion Matrix

# This script builds classification metrics manually.
# Confusion matrices reveal different prediction errors.
# Civil inspection labels make metrics concrete.

# Import small plotting and table libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Store true bridge inspection outcomes.
actual_labels = [
    "unsafe", "safe", "unsafe", "safe",
    "safe", "unsafe", "safe", "unsafe"
]

# Store model predictions for each inspection.
predicted_labels = [
    "unsafe", "safe", "safe", "safe",
    "unsafe", "unsafe", "safe", "unsafe"
]

# Validate matching list lengths before counting.
assert len(actual_labels) == len(predicted_labels)

# Define positive and negative class names.
positive_class = "unsafe"
negative_class = "safe"

# Count each confusion matrix category manually.
tp = sum(a == positive_class and p == positive_class for a, p in zip(actual_labels, predicted_labels))
tn = sum(a == negative_class and p == negative_class for a, p in zip(actual_labels, predicted_labels))
fp = sum(a == negative_class and p == positive_class for a, p in zip(actual_labels, predicted_labels))
fn = sum(a == positive_class and p == negative_class for a, p in zip(actual_labels, predicted_labels))

# Compute common classification metrics safely.
total = tp + tn + fp + fn
accuracy = (tp + tn) / total if total else 0
precision = tp / (tp + fp) if (tp + fp) else 0
recall = tp / (tp + fn) if (tp + fn) else 0

# Compute F1 score from precision and recall.
f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

# Print compact metric results for learners.
print(f"TP={tp}, TN={tn}, FP={fp}, FN={fn}")
print(f"Accuracy:  {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall:    {recall:.2f}")
print(f"F1 score:  {f1_score:.2f}")

# Build a labeled confusion matrix table.
matrix = pd.DataFrame(
    [[tn, fp], [fn, tp]],
    index=["Actual safe", "Actual unsafe"],
    columns=["Predicted safe", "Predicted unsafe"]
)

# Plot exactly one heatmap for visibility.
plt.figure(figsize=(5, 4))
sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix for Bridge Safety Classification")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.tight_layout()
plt.show()

```

### **2.2. Precision**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1783614248" width="250">



>* Measures correctness of positive predictions
>* Useful when false alarms are costly

>* Precision compares true positives to predicted positives
>* High precision means few false positives

>* High precision may miss subtle positives
>* Use precision when false positives matter



```
#@title Python Code - Precision

# Precision measures reliability of positive predictions.
# False positives reduce precision in classification.
# This example uses bridge inspection labels.

# Define actual bridge defect labels.
actual_labels = [1, 0, 1, 0, 1, 0, 0, 1]

# Define predicted bridge defect labels.
predicted_labels = [1, 1, 1, 0, 0, 0, 1, 1]

# Validate both lists before comparing.
assert len(actual_labels) == len(predicted_labels)

# Count true positive predictions manually.
true_positive = sum(
    1 for actual, predicted in zip(actual_labels, predicted_labels)
    if actual == 1 and predicted == 1
)

# Count false positive predictions manually.
false_positive = sum(
    1 for actual, predicted in zip(actual_labels, predicted_labels)
    if actual == 0 and predicted == 1
)

# Compute precision with safe division.
positive_predictions = true_positive + false_positive
precision = true_positive / positive_predictions if positive_predictions else 0

# Print compact teaching results.
print(f"True positives: {true_positive}")
print(f"False positives: {false_positive}")
print(f"Positive predictions: {positive_predictions}")
print(f"Precision: {precision:.2f}")
print("Interpretation: predicted defects were correct this often.")

```

### **2.3. F1 Score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1783614250" width="250">



>* Combines precision and recall performance
>* Useful for imbalanced important positive classes

>* Precision and recall often trade off
>* High F1 needs both to be strong

>* Balances false alarms and missed positives
>* Use alongside context-specific priorities



```
#@title Python Code - F1 Score

# This example computes classification F1 score.
# Precision and recall are balanced together.
# No machine learning library is needed.

# Store actual bridge inspection labels.
actual_labels = [1, 0, 1, 1, 0, 0, 1, 0]

# Store predicted bridge inspection labels.
predicted_labels = [1, 0, 0, 1, 0, 1, 1, 0]

# Check both lists have matching lengths.
assert len(actual_labels) == len(predicted_labels), "Label counts must match"

# Count true positives for damaged bridges.
true_positive = sum(1 for a, p in zip(actual_labels, predicted_labels) if a == 1 and p == 1)

# Count false positives for damaged bridges.
false_positive = sum(1 for a, p in zip(actual_labels, predicted_labels) if a == 0 and p == 1)

# Count false negatives for damaged bridges.
false_negative = sum(1 for a, p in zip(actual_labels, predicted_labels) if a == 1 and p == 0)

# Count true negatives for undamaged bridges.
true_negative = sum(1 for a, p in zip(actual_labels, predicted_labels) if a == 0 and p == 0)

# Compute precision with safe division.
precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) else 0

# Compute recall with safe division.
recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) else 0

# Compute F1 score with safe division.
f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

# Print the compact confusion matrix.
print("Confusion matrix: TP={}, FP={}, FN={}, TN={}".format(true_positive, false_positive, false_negative, true_negative))

# Print precision and recall values.
print("Precision = {:.2f}, Recall = {:.2f}".format(precision, recall))

# Print the final F1 score.
print("F1 score = {:.2f}".format(f1_score))

```

## **3. Generalization Issues**

### **3.1. Underfitting Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1783614239" width="250">



>* Too-simple models miss important data patterns
>* Training and testing errors both stay high

>* Similar poor scores can hide underfitting
>* High bias makes models too simplistic

>* Underfitting comes from limited learning capacity
>* Improve features, flexibility, training, and data



```
#@title Python Code - Underfitting Patterns

# This script demonstrates underfitting patterns.
# Both train and test errors stay high.
# A simple model misses curved behavior.

import numpy as np
import matplotlib.pyplot as plt

# Create deterministic civil engineering style data.
rng = np.random.default_rng(7)
x = np.linspace(0, 10, 40)

# True settlement curve is nonlinear.
noise = rng.normal(0, 1.2, x.size)
y = 2.0 + 0.8 * x + 0.45 * x**2 + noise

# Split data into training and testing sets.
train_mask = x <= 7.0
test_mask = x > 7.0

# Validate the split before fitting.
assert train_mask.sum() > 5
assert test_mask.sum() > 5

# Fit an overly simple constant model.
constant_prediction = float(np.mean(y[train_mask]))
train_pred = np.full(train_mask.sum(), constant_prediction)

test_pred = np.full(test_mask.sum(), constant_prediction)
train_true = y[train_mask]
test_true = y[test_mask]

# Compute mean squared errors manually.
train_mse = np.mean((train_true - train_pred) ** 2)
test_mse = np.mean((test_true - test_pred) ** 2)

# Compare errors and their gap.
error_gap = abs(test_mse - train_mse)
print(f"Training MSE: {train_mse:.2f}")
print(f"Testing MSE:  {test_mse:.2f}")
print(f"Error gap:    {error_gap:.2f}")
print("Pattern: both errors are high, so underfitting is likely.")

# Plot data and the underfit prediction.
plt.figure(figsize=(7, 4))
plt.scatter(x[train_mask], train_true, label="Training data")
plt.scatter(x[test_mask], test_true, label="Testing data")

# Draw the constant prediction line.
plt.axhline(constant_prediction, color="red", label="Too-simple model")
plt.xlabel("Load test position along beam, meters")
plt.ylabel("Measured settlement, millimeters")

# Add a clear teaching title.
plt.title("Underfitting: weak training and testing performance")
plt.legend()
plt.tight_layout()
plt.show()

```

### **3.2. Overfitting Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1783614243" width="250">



>* Memorizes noisy training details
>* High training performance, weak testing performance

>* Complex models may memorize noisy training quirks
>* High training accuracy may not generalize

>* Training improves while testing stalls or worsens
>* Deployment fails when models memorize past noise



```
#@title Python Code - Overfitting Patterns

# Compare training and testing errors.
# Overfitting shows a widening performance gap.
# This example uses simple polynomial fitting.

import numpy as np
import matplotlib.pyplot as plt

# Create repeatable synthetic civil engineering data.
rng = np.random.default_rng(7)

# Use traffic volume and bridge vibration response.
x_train = np.linspace(0, 10, 12)
y_train = 2.0 + 0.7 * x_train + rng.normal(0, 1.2, 12)

# Create unseen testing points from the same process.
x_test = np.linspace(0.5, 9.5, 12)
y_test = 2.0 + 0.7 * x_test + rng.normal(0, 1.2, 12)

# Validate matching input and output lengths.
assert x_train.size == y_train.size
assert x_test.size == y_test.size

# Define model complexities to compare.
degrees = [1, 3, 10]
train_mse_values = []
test_mse_values = []

# Fit each polynomial and measure errors.
for degree in degrees:
    coefficients = np.polyfit(x_train, y_train, degree)
    model = np.poly1d(coefficients)

    train_predictions = model(x_train)
    test_predictions = model(x_test)

    train_mse = np.mean((y_train - train_predictions) ** 2)
    test_mse = np.mean((y_test - test_predictions) ** 2)

    train_mse_values.append(train_mse)
    test_mse_values.append(test_mse)

# Print a compact comparison table.
print("Degree | Train MSE | Test MSE")
for degree, train_mse, test_mse in zip(degrees, train_mse_values, test_mse_values):
    print(f"{degree:6d} | {train_mse:9.2f} | {test_mse:8.2f}")

# Explain the overfitting pattern briefly.
print("Overfitting pattern: low training error but higher testing error.")

# Plot training and testing error patterns.
plt.figure(figsize=(7, 4))
plt.plot(degrees, train_mse_values, marker="o", label="Training MSE")
plt.plot(degrees, test_mse_values, marker="s", label="Testing MSE")

# Label the plot for interpretation.
plt.xlabel("Polynomial degree, representing model complexity")
plt.ylabel("Mean squared error")
plt.title("Overfitting Pattern: Training Versus Testing Error")
plt.legend()

# Display the single teaching plot.
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **3.3. Testing Performance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1783614242" width="250">



>* Test results estimate future model behavior
>* Similar train-test scores suggest real generalization

>* Poor train and test scores suggest underfitting
>* High train, low test scores suggest overfitting

>* Compare gaps and test performance levels
>* Judge readiness based on error consequences



```
#@title Python Code - Testing Performance

# Testing performance reveals model generalization.
# Train results alone can mislead.
# Compare gaps to diagnose fitting.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
np.random.seed(7)
area_sqft = np.linspace(500, 2500, 24)

# Add realistic rental price noise.
noise = np.random.normal(0, 120, area_sqft.size)
rent_usd = 900 + 1.15 * area_sqft + noise

# Split earlier listings for training.
train_x = area_sqft[:16]
train_y = rent_usd[:16]

# Keep later listings for testing.
test_x = area_sqft[16:]
test_y = rent_usd[16:]

# Validate matching training and testing sizes.
assert train_x.size == train_y.size
assert test_x.size == test_y.size

# Fit simple and flexible polynomial models.
degrees = [1, 3, 12]
labels = ["underfit", "balanced", "overfit"]

# Store errors for comparison.
train_rmse_values = []
test_rmse_values = []

# Define a small RMSE helper.
def rmse(actual_values, predicted_values):
    squared_errors = (actual_values - predicted_values) ** 2
    return float(np.sqrt(np.mean(squared_errors)))

# Evaluate each model complexity level.
for degree in degrees:
    coefficients = np.polyfit(train_x, train_y, degree)
    model = np.poly1d(coefficients)

    train_predictions = model(train_x)
    test_predictions = model(test_x)

    train_rmse_values.append(rmse(train_y, train_predictions))
    test_rmse_values.append(rmse(test_y, test_predictions))

# Print a compact diagnostic table.
print("Model complexity | Train RMSE | Test RMSE | Pattern")
for label, train_error, test_error in zip(labels, train_rmse_values, test_rmse_values):
    print(f"{label:16s} | {train_error:10.1f} | {test_error:9.1f}")

# Plot training and testing performance together.
plt.figure(figsize=(7, 4))
plt.plot(labels, train_rmse_values, marker="o", label="Training RMSE")
plt.plot(labels, test_rmse_values, marker="o", label="Testing RMSE")

# Lower RMSE means better prediction performance.
plt.ylabel("Rental price error, dollars")
plt.title("Testing performance exposes generalization issues")
plt.legend()

# Display the single teaching plot.
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
