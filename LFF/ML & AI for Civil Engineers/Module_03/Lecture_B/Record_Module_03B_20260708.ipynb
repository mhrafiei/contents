## **1. Regression Error Metrics**

### **1.1. Regression Residuals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1783556513" width="250">



>* Residuals show each prediction’s error size
>* They reveal where model mistakes occur

>* Residual signs show prediction direction
>* Consistent direction errors may reveal bias

>* Residual patterns reveal model fit problems
>* Residuals support MSE and error diagnosis



```
#@title Python Code - Regression Residuals

# Residuals show each prediction mistake.
# Positive values mean overprediction errors.
# Negative values mean underprediction errors.

# Store observed bridge deflections in millimeters.
actual_mm = [12.0, 18.5, 25.0, 31.5, 40.0]

# Store model predictions in millimeters.
predicted_mm = [11.0, 20.0, 23.5, 34.0, 38.0]

# Check both lists have matching lengths.
if len(actual_mm) != len(predicted_mm):
    raise ValueError("Actual and predicted lists must match.")

# Compute residuals as prediction minus actual.
residuals_mm = []
for predicted, actual in zip(predicted_mm, actual_mm):
    residuals_mm.append(predicted - actual)

# Print a compact residual table.
print("Case | Actual | Predicted | Residual")
for index, values in enumerate(zip(actual_mm, predicted_mm, residuals_mm), start=1):
    actual, predicted, residual = values
    print(f"{index:>4} | {actual:>6.1f} | {predicted:>9.1f} | {residual:>8.1f}")

# Count directional prediction mistakes.
over_count = sum(value > 0 for value in residuals_mm)
under_count = sum(value < 0 for value in residuals_mm)

# Summarize residual direction patterns.
print(f"Overpredictions: {over_count}; underpredictions: {under_count}")
print("Residuals are the raw errors used by later metrics.")

```

### **1.2. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1783556514" width="250">



>* MSE averages squared prediction errors
>* Lower MSE means closer predictions

>* Squaring stops opposite errors from canceling
>* Large mistakes receive stronger penalties

>* Compare, square, then average prediction errors
>* Lower MSE helps compare regression models



```
#@title Python Code - Mean Squared Error

# This example calculates regression prediction errors.
# Mean squared error rewards smaller mistakes.
# Squaring makes all residuals positive.

# Store actual concrete strengths in megapascals.
actual_strengths = [28.0, 32.5, 35.0, 40.0]

# Store predicted concrete strengths in megapascals.
predicted_strengths = [30.0, 31.0, 36.5, 37.0]

# Check both lists have matching lengths.
same_length = len(actual_strengths) == len(predicted_strengths)

# Stop early if paired values are missing.
if not same_length:
    raise ValueError("Actual and predicted lists must match.")

# Compute residuals by subtracting actual values.
errors = []
for actual, predicted in zip(actual_strengths, predicted_strengths):
    errors.append(predicted - actual)

# Square each residual to remove signs.
squared_errors = []
for error in errors:
    squared_errors.append(error ** 2)

# Average squared errors to get MSE.
mse_value = sum(squared_errors) / len(squared_errors)

# Show compact intermediate teaching results.
print("Actual strengths MPa:", actual_strengths)
print("Predicted strengths MPa:", predicted_strengths)
print("Residual errors MPa:", errors)
print("Squared errors MPa^2:", squared_errors)
print("Mean squared error:", round(mse_value, 3), "MPa^2")

```

### **1.3. Error Calculation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1783556516" width="250">



>* Compare predictions with actual values
>* Use error size, not direction

>* Squaring residuals prevents error cancellation
>* Large mistakes affect MSE more strongly

>* Average squared residuals for one error score
>* Compare models, but note squared units



```
#@title Python Code - Error Calculation

# Manual errors support regression metric learning.
# Tiny civil data keeps calculations clear.
# Squared residuals prevent cancellation problems.

# No extra installations are required.

# Store actual bridge deflections in millimeters.
actual_mm = [12.0, 18.0, 25.0, 30.0]

# Store predicted bridge deflections in millimeters.
predicted_mm = [10.0, 20.0, 22.0, 35.0]

# Validate matching observation counts before calculation.
if len(actual_mm) != len(predicted_mm):
    raise ValueError("Actual and predicted lists must match.")

# Compute residuals as actual minus predicted.
residuals = []
for actual, predicted in zip(actual_mm, predicted_mm):
    residuals.append(actual - predicted)

# Square each residual to remove signs.
squared_errors = []
for residual in residuals:
    squared_errors.append(residual ** 2)

# Average squared errors to get MSE.
count_values = len(squared_errors)
if count_values == 0:
    raise ValueError("At least one observation is required.")

# Calculate the manual mean squared error.
mse_value = sum(squared_errors) / count_values

# Show compact row level error details.
print("Actual, predicted, residual, squared error")
for row in range(count_values):
    print(actual_mm[row], predicted_mm[row], residuals[row], squared_errors[row])

# Show the final regression error summary.
print("Manual MSE:", round(mse_value, 2), "square millimeters")

```

## **2. Classification Metrics**

### **2.1. Confusion Matrix**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1783556507" width="250">



>* Compares predicted labels with true labels
>* Shows true and false prediction types

>* Different errors can have different consequences
>* Confusion matrices reveal costly mistake patterns

>* Foundation for key classification metrics
>* Reveals context-specific model decision quality



```
#@title Python Code - Confusion Matrix

# Build a small classification example.
# Compare true labels with predictions.
# Count each confusion matrix cell.

# Define observed and predicted bridge inspection labels.
true_labels = [1, 0, 1, 1, 0, 0, 1, 0]
pred_labels = [1, 0, 0, 1, 0, 1, 1, 0]
positive_name = "needs repair"

# Check both lists have matching lengths.
if len(true_labels) != len(pred_labels):
    raise ValueError("Label lists must match.")

# Count the four confusion matrix outcomes.
tp = sum(t == 1 and p == 1 for t, p in zip(true_labels, pred_labels))
tn = sum(t == 0 and p == 0 for t, p in zip(true_labels, pred_labels))
fp = sum(t == 0 and p == 1 for t, p in zip(true_labels, pred_labels))
fn = sum(t == 1 and p == 0 for t, p in zip(true_labels, pred_labels))

# Store the matrix using common row and column order.
confusion_matrix = [[tn, fp], [fn, tp]]
total_cases = len(true_labels)
correct_cases = tp + tn

# Print a compact confusion matrix summary.
print("Positive class:", positive_name)
print("Confusion matrix [[TN, FP], [FN, TP]]:", confusion_matrix)
print("True negatives:", tn, "False positives:", fp)
print("False negatives:", fn, "True positives:", tp)

# Compute accuracy safely from all cases.
accuracy = correct_cases / total_cases if total_cases else 0
print("Accuracy:", round(accuracy, 3))

```

### **2.2. Precision**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1783556509" width="250">



>* Precision measures correctness of positive predictions
>* High precision reduces harmful false positives

>* Precision matters when false positives are costly
>* It measures trust in positive predictions

>* Balance precision with recall and other metrics
>* High precision may miss important positives



```
#@title Python Code - Precision

# Precision measures trust in positive predictions.
# This example uses bridge inspection alerts.
# We compute precision from confusion counts.

# Store actual bridge risk labels.
actual_labels = [1, 0, 1, 0, 1, 0, 0, 1]

# Store model predicted risk alerts.
predicted_labels = [1, 1, 1, 0, 0, 0, 1, 1]

# Validate equal list lengths before counting.
same_length = len(actual_labels) == len(predicted_labels)
assert same_length, "Actual and predicted lists must match."

# Count true positive bridge alerts.
true_positive = sum(
    actual == 1 and predicted == 1
    for actual, predicted in zip(actual_labels, predicted_labels)
)

# Count false positive bridge alerts.
false_positive = sum(
    actual == 0 and predicted == 1
    for actual, predicted in zip(actual_labels, predicted_labels)
)

# Compute precision with safe division.
positive_predictions = true_positive + false_positive
precision = true_positive / positive_predictions if positive_predictions else 0

# Print compact teaching results.
print("True positives:", true_positive)
print("False positives:", false_positive)
print("Positive predictions:", positive_predictions)
print("Precision:", round(precision, 3))
print("Meaning: trusted alerts among all predicted risk alerts.")

```

### **2.3. F1 Score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1783556511" width="250">



>* Combines precision and recall for positives
>* Useful when accuracy hides class imbalance

>* Precision checks positive prediction reliability
>* F1 balances precision and recall trade-offs

>* Balance false positives and false negatives
>* Compare with context and other metrics



```
#@title Python Code - F1 Score

# This example computes F1 from classification counts.
# It uses a small civil engineering inspection dataset.
# No machine learning library is needed here.

# Store true labels for bridge crack inspections.
actual_labels = [1, 0, 1, 1, 0, 0, 1, 0]

# Store predicted labels from a simple classifier.
predicted_labels = [1, 0, 0, 1, 0, 1, 1, 0]

# Check both lists have matching lengths.
same_length = len(actual_labels) == len(predicted_labels)

# Stop early if labels cannot be compared.
assert same_length, "Label lists must have equal length."

# Count true positives for the positive class.
true_positive = sum(a == 1 and p == 1 for a, p in zip(actual_labels, predicted_labels))

# Count false positives for the positive class.
false_positive = sum(a == 0 and p == 1 for a, p in zip(actual_labels, predicted_labels))

# Count false negatives for the positive class.
false_negative = sum(a == 1 and p == 0 for a, p in zip(actual_labels, predicted_labels))

# Count true negatives for the negative class.
true_negative = sum(a == 0 and p == 0 for a, p in zip(actual_labels, predicted_labels))

# Compute precision with safe zero handling.
precision_denominator = true_positive + false_positive
precision = true_positive / precision_denominator if precision_denominator else 0

# Compute recall with safe zero handling.
recall_denominator = true_positive + false_negative
recall = true_positive / recall_denominator if recall_denominator else 0

# Combine precision and recall into F1.
f1_denominator = precision + recall
f1_score = 2 * precision * recall / f1_denominator if f1_denominator else 0

# Print compact results for interpretation.
print("Confusion matrix counts: TP=", true_positive, "FP=", false_positive)
print("Confusion matrix counts: FN=", false_negative, "TN=", true_negative)
print("Precision:", round(precision, 3), "Recall:", round(recall, 3))
print("F1 score:", round(f1_score, 3))
print("F1 is high only when precision and recall are balanced.")

```

## **3. Generalization Issues**

### **3.1. Underfitting Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1783556501" width="250">



>* Too simple to learn key patterns
>* Poor on both training and testing

>* Weak training performance suggests underfitting
>* Limited features miss important prediction patterns

>* Both training and testing scores stay low
>* Improve capacity, features, training, or assumptions



```
#@title Python Code - Underfitting Patterns

# Underfitting shows weak learning on all data.
# Simple models can miss important engineering patterns.
# Compare train and test errors together.

import numpy as np
import matplotlib.pyplot as plt

# Create a small curved settlement dataset.
rng = np.random.default_rng(7)
x = np.linspace(0, 10, 24)

# Settlement depends nonlinearly on load.
noise = rng.normal(0, 1.2, x.size)
y = 4 + 1.5 * x + 0.9 * x**2 + noise

# Split data into training and testing parts.
train_mask = np.arange(x.size) % 3 != 0
test_mask = np.logical_not(train_mask)

# Use an overly simple constant model.
constant_prediction = np.mean(y[train_mask])
train_pred = np.full(train_mask.sum(), constant_prediction)

test_pred = np.full(test_mask.sum(), constant_prediction)
train_mse = np.mean((y[train_mask] - train_pred) ** 2)
test_mse = np.mean((y[test_mask] - test_pred) ** 2)

# Print a compact diagnostic summary.
print("Underfitting diagnostic for settlement prediction")
print(f"Training MSE: {train_mse:.1f}")
print(f"Testing MSE:  {test_mse:.1f}")
print("Pattern: both errors are high and similar.")

# Plot data and the too-simple model.
plt.figure(figsize=(7, 4))
plt.scatter(x[train_mask], y[train_mask], label="training points")
plt.scatter(x[test_mask], y[test_mask], label="testing points")

# Show the constant prediction line.
plt.axhline(constant_prediction, color="red", label="constant model")
plt.xlabel("Applied load, tons")
plt.ylabel("Measured settlement, millimeters")
plt.title("Underfitting: model is too simple")

# Keep the plot readable for beginners.
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.2. Overfitting Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1783556505" width="250">



>* Memorizes training details instead of general patterns
>* High training performance, weak testing performance

>* Growing train-test gaps signal overfitting
>* Models may learn noise, not general patterns

>* High training scores can mislead.
>* Weak testing scores require careful investigation.



```
#@title Python Code - Overfitting Patterns

# Compare training and testing performance patterns.
# Overfitting shows excellent training but weak testing.
# This example uses simple polynomial regression.

import numpy as np
import matplotlib.pyplot as plt

# Set a seed for repeatable results.
rng = np.random.default_rng(7)

# Create small civil engineering style data.
x_train = np.linspace(0, 10, 12)
y_train = 3 + 2 * x_train + rng.normal(0, 2.0, 12)

# Create unseen testing data from same relationship.
x_test = np.linspace(0.5, 9.5, 12)
y_test = 3 + 2 * x_test + rng.normal(0, 2.0, 12)

# Validate matching input and output sizes.
assert x_train.size == y_train.size
assert x_test.size == y_test.size

# Define mean squared error manually.
def mean_squared_error(actual, predicted):
    errors = actual - predicted
    return float(np.mean(errors ** 2))

# Fit a simple model and complex model.
linear_fit = np.polyfit(x_train, y_train, 1)
complex_fit = np.polyfit(x_train, y_train, 10)

# Predict training and testing values.
linear_train = np.polyval(linear_fit, x_train)
linear_test = np.polyval(linear_fit, x_test)

# Predict using the complex polynomial.
complex_train = np.polyval(complex_fit, x_train)
complex_test = np.polyval(complex_fit, x_test)

# Calculate manual MSE values.
linear_train_mse = mean_squared_error(y_train, linear_train)
linear_test_mse = mean_squared_error(y_test, linear_test)

# Calculate complex model MSE values.
complex_train_mse = mean_squared_error(y_train, complex_train)
complex_test_mse = mean_squared_error(y_test, complex_test)

# Print a compact performance comparison.
print("Model comparison using manual MSE")
print(f"Linear train MSE:  {linear_train_mse:.2f}")
print(f"Linear test MSE:   {linear_test_mse:.2f}")
print(f"Complex train MSE: {complex_train_mse:.2f}")
print(f"Complex test MSE:  {complex_test_mse:.2f}")
print("Overfitting pattern: low training error, high testing error.")

# Prepare smooth curves for one plot.
x_curve = np.linspace(0, 10, 200)
linear_curve = np.polyval(linear_fit, x_curve)
complex_curve = np.polyval(complex_fit, x_curve)

# Plot data and fitted curves.
plt.figure(figsize=(7, 4))
plt.scatter(x_train, y_train, label="Training data")
plt.scatter(x_test, y_test, label="Testing data")

# Add simple and complex model curves.
plt.plot(x_curve, linear_curve, label="Simple linear model")
plt.plot(x_curve, complex_curve, label="Complex polynomial model")
plt.xlabel("Beam span length, meters")
plt.ylabel("Estimated cost, thousand dollars")

# Finish the plot with readable labels.
plt.title("Overfitting Pattern: Train Fit Versus Test Performance")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Testing Performance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1783556503" width="250">



>* Test results show real-world generalization
>* Weak tests reveal fitting or data issues

>* Poor train and test scores suggest underfitting
>* Large train-test gaps suggest overfitting

>* Single test scores can mislead
>* Reliable tests match real future conditions



```
#@title Python Code - Testing Performance

# Testing performance reveals model generalization.
# Train scores alone can mislead.
# Compare train and test errors.

import numpy as np
import matplotlib.pyplot as plt

# Use deterministic synthetic civil data.
rng = np.random.default_rng(7)

# Create bridge span lengths in meters.
span_m = np.linspace(10, 100, 30)

# Create nonlinear inspection cost values.
true_cost = 50 + 2.2 * span_m + 0.035 * span_m**2

# Add realistic measurement noise.
noise = rng.normal(0, 45, span_m.size)

# Combine signal and noise.
cost_k = true_cost + noise

# Split early observations for training.
train_x = span_m[:20]
train_y = cost_k[:20]

# Reserve later observations for testing.
test_x = span_m[20:]
test_y = cost_k[20:]

# Validate matching training shapes.
assert train_x.shape == train_y.shape

# Validate matching testing shapes.
assert test_x.shape == test_y.shape

# Define mean squared error manually.
def mse_manual(actual, predicted):
    errors = actual - predicted
    squared_errors = errors ** 2
    return float(np.mean(squared_errors))

# Fit simple and complex polynomials.
degrees = [1, 3, 12]

# Store results for comparison.
results = []

# Evaluate each model silently.
for degree in degrees:
    coeffs = np.polyfit(train_x, train_y, degree)
    train_pred = np.polyval(coeffs, train_x)
    test_pred = np.polyval(coeffs, test_x)
    results.append((degree, mse_manual(train_y, train_pred), mse_manual(test_y, test_pred)))

# Print compact performance summary.
print("Model degree | Train MSE | Test MSE | Pattern")

# Interpret train versus test performance.
for degree, train_mse, test_mse in results:
    if train_mse > 9000 and test_mse > 9000:
        pattern = "underfitting"
    elif test_mse > train_mse * 4:
        pattern = "overfitting"
    else:
        pattern = "better generalization"
    print(f"{degree:12d} | {train_mse:9.0f} | {test_mse:8.0f} | {pattern}")

# Plot data and model curves.
grid_x = np.linspace(10, 100, 200)

# Prepare one comparison figure.
plt.figure(figsize=(8, 5))

# Show training and testing points.
plt.scatter(train_x, train_y, label="training data", color="tab:blue")
plt.scatter(test_x, test_y, label="testing data", color="tab:orange")

# Draw each fitted curve.
for degree in degrees:
    coeffs = np.polyfit(train_x, train_y, degree)
    grid_y = np.polyval(coeffs, grid_x)
    plt.plot(grid_x, grid_y, label=f"degree {degree}")

# Label the teaching plot.
plt.xlabel("Bridge span length, meters")
plt.ylabel("Inspection cost, thousand dollars")
plt.title("Testing performance exposes generalization issues")
plt.legend()

# Display the single plot.
plt.show()

```
