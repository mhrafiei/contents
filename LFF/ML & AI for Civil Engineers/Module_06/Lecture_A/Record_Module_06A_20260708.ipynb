## **1. Unsupervised Pattern Discovery**

### **1.1. No Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_01.jpg?v=1783562989" width="250">



>* Unsupervised data has no known answers
>* Methods find groups, patterns, or unusual cases

>* No labels mean no simple answer key
>* Patterns need later engineering interpretation

>* Useful for exploring unlabeled engineering data
>* Clusters need cautious engineering interpretation



```
#@title Python Code - No Labels

# Demonstrate clustering without known engineering labels.
# Compare measurements only, not answer keys.
# Use tiny synthetic building operation data.

import numpy as np
import matplotlib.pyplot as plt

# Create repeatable unlabeled building measurements.
rng = np.random.default_rng(seed=7)

# Define three hidden operating patterns.
centers = np.array([[35, 18], [70, 24], [95, 31]])

# Generate small noisy observations around patterns.
points = np.vstack([center + rng.normal(0, 4, (8, 2)) for center in centers])

# Confirm the data table has two measurements.
assert points.shape == (24, 2)

# Choose starting centroids manually from observations.
centroids = points[[0, 10, 20]].copy()

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(points[:, None, :] - centroids[None, :, :], axis=2)
    groups = np.argmin(distances, axis=1)
    centroids = np.array([points[groups == k].mean(axis=0) for k in range(3)])

# Print concise interpretation reminders.
print("Unsupervised learning: measurements are given, labels are not.")
print("K-means found groups using electricity and temperature only.")
print("Cluster numbers are not physical truths by themselves.")
print("Engineers must interpret groups using system knowledge.")

# Plot measurements and discovered clusters.
plt.figure(figsize=(6, 4))
plt.scatter(points[:, 0], points[:, 1], c=groups, cmap="viridis", s=70)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)
plt.xlabel("Daily electricity demand, kWh")
plt.ylabel("Average indoor temperature, °C")
plt.title("No labels: clusters discovered from measurements")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Prediction Versus Discovery**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_02.jpg?v=1783562991" width="250">



>* Supervised models predict known target outcomes
>* Unsupervised methods reveal unlabeled data patterns

>* Supervised models have clear evaluation targets
>* Clusters are exploratory, variable-dependent groupings

>* Clustering can reveal useful hidden operating patterns
>* Engineers must validate clusters before using them



```
#@title Python Code - Prediction Versus Discovery

# Compare prediction with discovery.
# Use tiny engineering measurements.
# Plot labels versus clusters.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make the example deterministic.
rng = np.random.default_rng(7)

# Create two operating regimes.
low_flow = rng.normal([22, 4], [2, 0.5], (10, 2))
high_flow = rng.normal([38, 9], [2, 0.7], (10, 2))

# Combine unlabeled sensor measurements.
X = np.vstack([low_flow, high_flow])
true_labels = np.array([0] * 10 + [1] * 10)

# Validate the small data shape.
assert X.shape == (20, 2)
assert true_labels.size == X.shape[0]

# Choose starting centroids manually.
centroids = np.array([[20.0, 4.0], [40.0, 9.0]])

# Run a few transparent updates.
for step in range(4):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    cluster_ids = np.argmin(distances, axis=1)
    centroids = np.array([X[cluster_ids == k].mean(axis=0) for k in range(2)])

# Summarize prediction versus discovery.
print("Supervised prediction would need known target labels.")
print("K-means used only flow and vibration measurements.")
print("Clusters are discovered groups, not guaranteed fault types.")
print(f"Final centroids: {np.round(centroids, 2).tolist()}")

# Plot discovered clusters beside known labels.
fig, axes = plt.subplots(1, 2, figsize=(9, 4))

# Show labels as supervised information.
axes[0].scatter(X[:, 0], X[:, 1], c=true_labels, cmap="coolwarm", s=60)
axes[0].set_title("Known labels for prediction")
axes[0].set_xlabel("Flow rate, gallons per minute")
axes[0].set_ylabel("Vibration, millimeters per second")

# Show clusters as discovered structure.
axes[1].scatter(X[:, 0], X[:, 1], c=cluster_ids, cmap="viridis", s=60)
axes[1].scatter(centroids[:, 0], centroids[:, 1], c="black", marker="X", s=120)
axes[1].set_title("Discovered k-means clusters")
axes[1].set_xlabel("Flow rate, gallons per minute")

# Keep the single plot readable.
plt.tight_layout()
plt.show()

```

### **1.3. Engineering Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_03.jpg?v=1783562987" width="250">



>* Unsupervised patterns need engineering context
>* Clusters suggest discoveries, not final answers

>* Check clusters against engineering context
>* Consider real causes before labeling groups

>* Clusters may reflect data artifacts
>* Use engineering evidence to verify meaning



```
#@title Python Code - Engineering Interpretation

# Interpret clusters using engineering context.
# Unsupervised groups need cautious explanations.
# Small pump data supports discussion.

# No extra installations are required.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create small pump operating observations.
pump_names = np.array(["P1", "P2", "P3", "P4", "P5", "P6"])
flow_gpm = np.array([510, 495, 520, 300, 315, 305])
power_kw = np.array([42, 41, 44, 39, 40, 38])
zone = np.array(["low", "low", "low", "high", "high", "high"])

# Standardize variables before distance calculations.
X = np.column_stack((flow_gpm, power_kw))
means = X.mean(axis=0)
scales = X.std(axis=0)
X_scaled = (X - means) / scales

# Validate the tiny teaching dataset.
assert X_scaled.shape == (6, 2)
assert np.all(scales > 0)

# Choose two starting centroids manually.
centroids = X_scaled[[0, 3]].copy()

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(X_scaled[:, None, :] - centroids[None, :, :], axis=2)
    labels = distances.argmin(axis=1)
    centroids = np.array([X_scaled[labels == k].mean(axis=0) for k in range(2)])

# Summarize clusters without overclaiming causes.
for k in range(2):
    members = pump_names[labels == k]
    zones = zone[labels == k]
    print(f"Cluster {k}: pumps {list(members)}, zones {list(zones)}")

# Print the engineering interpretation warning.
print("Clusters separate operating context, not proven pump health.")
print("Check elevation, demand, sensors, and maintenance records.")

# Plot clusters against physical variables.
colors = np.where(labels == 0, "tab:blue", "tab:orange")
plt.figure(figsize=(6, 4))
plt.scatter(flow_gpm, power_kw, c=colors, s=90)

# Label each pump for interpretation.
for name, x_value, y_value in zip(pump_names, flow_gpm, power_kw):
    plt.text(x_value + 3, y_value + 0.1, name)

# Add context labels and title.
plt.xlabel("Flow rate, gallons per minute")
plt.ylabel("Power consumption, kilowatts")
plt.title("K-means clusters need engineering interpretation")
plt.grid(True, alpha=0.3)

# Display the single teaching plot.
plt.show()

```

## **2. K Means Steps**

### **2.1. Initialize Centroids**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_01.jpg?v=1783562992" width="250">



>* Choose starting centers before assigning data points
>* Initial centers guide the clustering path

>* Spread starting centroids across observed data
>* Use domain knowledge; avoid poor placement

>* Starting centroids affect final cluster results
>* Compare choices using plots and judgment



```
#@title Python Code - Initialize Centroids

# Manual centroids start the k-means process.
# Small vibration data represent machinery conditions.
# We choose spread-out starting center locations.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Store vibration amplitude and frequency measurements.
data = np.array([
    [0.8, 18.0], [1.0, 20.0], [1.2, 19.0],

    [2.8, 35.0], [3.0, 37.0], [3.2, 36.0],
    [5.0, 55.0], [5.3, 57.0], [5.6, 56.0]
])

# Validate the two-dimensional engineering measurement table.
assert data.shape == (9, 2), "Expected nine two-feature observations."

# Manually initialize three spread-out centroid locations.
centroids = np.array([
    [0.9, 19.0],
    [3.0, 36.0],
    [5.4, 56.0]
])

# Validate the chosen centroid table shape.
assert centroids.shape == (3, 2), "Expected three two-feature centroids."

# Print a compact summary of the initialization.
print("Data points:", data.shape[0])
print("Manual centroids:", centroids.shape[0])
print("Features: amplitude in mm, frequency in Hz")

# Create one plot showing data and initial centroids.
plt.figure(figsize=(7, 5))
plt.scatter(data[:, 0], data[:, 1], label="measurements")

# Mark initial centroids with large red crosses.
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    marker="X", s=180, c="red", label="initial centroids"
)

# Label axes using engineering measurement units.
plt.xlabel("Vibration amplitude (mm)")
plt.ylabel("Dominant frequency (Hz)")
plt.title("Manual Initialization of K-Means Centroids")

# Add grid and legend for interpretation.
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

```

### **2.2. Assign Points**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_02.jpg?v=1783562996" width="250">



>* Assign each point to its nearest centroid
>* Create initial groups for later improvement

>* Closeness depends on chosen data features
>* Scale features to avoid misleading assignments

>* Cluster assignments can change after centroid updates
>* K-means improves through repeated reassignment



```
#@title Python Code - Assign Points

# K-means assigns points to nearest centroids.
# Distances compare engineering feature profiles.
# Assignments are temporary during iteration.

import numpy as np
import matplotlib.pyplot as plt

# Create small engineering observations.
points = np.array([
    [1.0, 2.0], [1.5, 1.8],
    [5.0, 8.0], [6.0, 8.5],
    [9.0, 2.0], [8.5, 3.0]
])

# Choose initial centroids manually.
centroids = np.array([
    [1.0, 2.0], [5.0, 8.0],
    [9.0, 2.0]
])

# Validate compatible two-dimensional shapes.
assert points.shape[1] == centroids.shape[1]
assert len(points) >= len(centroids)

# Compute distances from points to centroids.
differences = points[:, None, :] - centroids[None, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))
assignments = np.argmin(distances, axis=1)

# Print compact assignment results.
print("Point assignments to nearest centroid:")
for index, cluster in enumerate(assignments):
    print(f"Point {index + 1} -> Cluster {cluster + 1}")

# Plot points colored by assigned cluster.
colors = ["tab:blue", "tab:orange", "tab:green"]
for cluster_id in range(len(centroids)):
    mask = assignments == cluster_id
    plt.scatter(points[mask, 0], points[mask, 1], color=colors[cluster_id])

# Plot centroids as large black markers.
plt.scatter(centroids[:, 0], centroids[:, 1], c="black", marker="X", s=160)
plt.xlabel("Vibration level, millimeters per second")
plt.ylabel("Operating temperature, degrees Celsius")
plt.title("K-means assignment step")
plt.grid(True)

plt.show()

```

### **2.3. Centroid Update**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_03.jpg?v=1783562994" width="250">



>* Move centroids to assigned points’ average
>* Updated centers improve cluster grouping

>* Updates move rough centroids toward assigned points
>* Repeated updates improve the next assignments

>* Alternate assignments with careful centroid recalculation
>* Handle empty clusters; stop when stable



```
#@title Python Code - Centroid Update

# Demonstrate centroid updating with pump sensor data.
# Use only NumPy and one simple plot.
# Keep calculations visible for beginner learners.

import numpy as np
import matplotlib.pyplot as plt

# Store vibration and temperature measurements.
points = np.array([
    [0.18, 62.0],
    [0.22, 64.0],
    [0.25, 63.0],
    [0.70, 82.0],
    [0.75, 85.0],
    [0.80, 83.0],
])

# Choose starting centroids manually.
centroids = np.array([
    [0.20, 61.0],
    [0.78, 86.0],
])

# Validate the small teaching dataset.
assert points.shape == (6, 2)
assert centroids.shape == (2, 2)

# Compute distances from points to centroids.
differences = points[:, None, :] - centroids[None, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))
assignments = np.argmin(distances, axis=1)

# Update each centroid using assigned points.
updated_centroids = centroids.copy()
for cluster_id in range(len(centroids)):
    cluster_points = points[assignments == cluster_id]
    if len(cluster_points) > 0:
        updated_centroids[cluster_id] = cluster_points.mean(axis=0)

# Print a compact summary of movement.
print("Manual k-means centroid update")
print("Cluster assignments:", assignments.tolist())
print("Old centroids:", np.round(centroids, 2).tolist())
print("Updated centroids:", np.round(updated_centroids, 2).tolist())

# Plot points and centroid movement.
colors = np.array(["tab:blue", "tab:orange"])
plt.figure(figsize=(7, 4))
plt.scatter(points[:, 0], points[:, 1], c=colors[assignments], s=80)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=140, c="black")
plt.scatter(updated_centroids[:, 0], updated_centroids[:, 1], marker="*", s=180, c="red")

# Draw arrows from old to updated centroids.
for old, new in zip(centroids, updated_centroids):
    plt.arrow(old[0], old[1], new[0] - old[0], new[1] - old[1],
              length_includes_head=True, head_width=0.015, color="gray")

# Label the engineering measurements.
plt.xlabel("Vibration amplitude, inches per second")
plt.ylabel("Operating temperature, degrees Fahrenheit")
plt.title("Centroid update after one assignment step")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

## **3. Interpreting Clusters**

### **3.1. Cluster Context**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_03_01.jpg?v=1783562981" width="250">



>* Clusters reflect chosen variables and scaling
>* Use engineering context before assigning meaning

>* Plots reveal cluster shape and separation
>* Use context before assigning engineering meaning

>* Missing variables can hide engineering causes
>* Test cluster stability before assigning meaning



```
#@title Python Code - Cluster Context

# Clusters need engineering context.
# K-means labels are not explanations.
# Plots support cautious interpretation.

import numpy as np
import matplotlib.pyplot as plt

# Create deterministic pump operating data.
rng = np.random.default_rng(7)

# Define two hidden operating contexts.
loads = rng.normal(80, 6, 18)
flows = rng.normal(210, 12, 18)

# Add similar measurements from another context.
loads = np.r_[loads, rng.normal(82, 6, 18)]
flows = np.r_[flows, rng.normal(208, 12, 18)]

# Store context not used by clustering.
context = np.array(["old wing"] * 18 + ["new wing"] * 18)

# Combine variables used by k-means.
X = np.column_stack([loads, flows])

# Validate the small teaching dataset.
assert X.shape == (36, 2)

# Scale variables before distance calculations.
means = X.mean(axis=0)
stds = X.std(axis=0)
Z = (X - means) / stds

# Choose manual starting centroids.
centroids = Z[[3, 25]].copy()

# Run a few transparent k-means updates.
for step in range(6):
    distances = ((Z[:, None, :] - centroids[None, :, :]) ** 2).sum(axis=2)
    labels = distances.argmin(axis=1)
    centroids = np.vstack([Z[labels == k].mean(axis=0) for k in range(2)])

# Count contexts inside each cluster.
for k in range(2):
    old_count = np.sum((labels == k) & (context == "old wing"))
    new_count = np.sum((labels == k) & (context == "new wing"))
    print(f"Cluster {k}: old wing={old_count}, new wing={new_count}")

# Print the main interpretation warning.
print("Similar clusters can mix different engineering contexts.")
print("Use plots and system knowledge before naming clusters.")

# Plot clusters and hidden context markers.
markers = {"old wing": "o", "new wing": "s"}
colors = np.array(["tab:blue", "tab:orange"])[labels]

# Draw each context with cluster colors.
for name, marker in markers.items():
    mask = context == name
    plt.scatter(loads[mask], flows[mask], c=colors[mask], marker=marker, label=name)

# Add labels and a cautious title.
plt.xlabel("Pump load, percent")
plt.ylabel("Flow rate, gallons per minute")
plt.title("K-means clusters need operational context")

# Show legend and final plot.
plt.legend(title="Known context")
plt.show()

```

### **3.2. Engineering Context**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_03_02.jpg?v=1783562985" width="250">



>* K-means finds patterns, not physical meaning
>* Use engineering knowledge to validate clusters

>* Use plots as clues, not proof
>* Check engineering context before interpreting clusters

>* Avoid labeling clusters without supporting evidence
>* Use clusters to guide deeper investigation



```
#@title Python Code - Engineering Context

# Clusters need engineering context.
# Plots support cautious interpretation.
# K-means labels are not diagnoses.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create deterministic building energy observations.
rng = np.random.default_rng(7)
small_area = rng.normal(18000, 1200, 12)
large_area = rng.normal(52000, 2500, 12)

# Simulate energy use with occupancy effects.
small_energy = rng.normal(420, 35, 12)
large_energy = rng.normal(980, 70, 12)

# Combine features for manual k-means.
area = np.concatenate([small_area, large_area])
energy = np.concatenate([small_energy, large_energy])
X = np.column_stack([area, energy])

# Validate the tiny teaching dataset.
assert X.shape == (24, 2)
assert np.isfinite(X).all()

# Scale variables before distance calculations.
means = X.mean(axis=0)
stds = X.std(axis=0)
Xs = (X - means) / stds

# Choose two starting centroids manually.
centroids = Xs[[2, 20]].copy()
labels = np.zeros(Xs.shape[0], dtype=int)

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(Xs[:, None, :] - centroids[None, :, :], axis=2)
    labels = distances.argmin(axis=1)
    centroids = np.array([Xs[labels == k].mean(axis=0) for k in range(2)])

# Summarize clusters without overclaiming causes.
for k in range(2):
    count = int((labels == k).sum())
    mean_area = area[labels == k].mean()
    mean_energy = energy[labels == k].mean()
    print(f"Cluster {k}: {count} buildings, {mean_area:.0f} ft^2, {mean_energy:.0f} MWh")

# Print cautious engineering interpretation guidance.
print("Interpretation: clusters suggest patterns, not automatic efficiency labels.")
print("Check occupancy, climate, equipment age, and controls before naming groups.")

# Plot clusters using original engineering units.
plt.figure(figsize=(7, 4.5))
plt.scatter(area, energy, c=labels, cmap="viridis", s=70, edgecolor="black")
plt.xlabel("Building floor area (ft^2)")
plt.ylabel("Annual energy use (MWh)")
plt.title("K-means clusters need engineering context")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **3.3. Practical Engineering Judgment**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_03_03.jpg?v=1783562983" width="250">



>* Clusters suggest patterns, not final answers
>* Check meanings with engineering context

>* Plots can hide important data dimensions
>* Check data quality before trusting clusters

>* Validate clusters with engineering evidence
>* Use clusters as guidance, not expertise



```
#@title Python Code - Practical Engineering Judgment

# Clusters need engineering judgment.
# Plots suggest hypotheses, not diagnoses.
# Context can change cluster meaning.

import numpy as np
import matplotlib.pyplot as plt

# Create small vibration and temperature observations.
rng = np.random.default_rng(7)
normal = rng.normal([2.0, 55.0], [0.25, 3.0], (10, 2))

# Add similar points from different operating speeds.
high_speed = rng.normal([3.0, 70.0], [0.25, 3.0], (10, 2))
wear = rng.normal([4.2, 82.0], [0.30, 4.0], (10, 2))

# Combine measurements into one engineering dataset.
X = np.vstack([normal, high_speed, wear])
assert X.shape == (30, 2)

# Store context that clustering cannot see.
context = np.array(["normal"] * 10 + ["high speed"] * 10 + ["bearing wear"] * 10)
centroids = X[[0, 12, 25]].copy()

# Run a tiny manual k-means procedure.
for step in range(5):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([X[labels == k].mean(axis=0) for k in range(3)])

# Summarize clusters without overclaiming causes.
print("Manual k-means formed three groups from two measured variables.")
print("Cluster labels are patterns, not automatic fault diagnoses.")
print("Engineering context should be checked before naming clusters.")

# Count hidden operating contexts inside each cluster.
for k in range(3):
    names, counts = np.unique(context[labels == k], return_counts=True)
    summary = ", ".join([f"{n}: {c}" for n, c in zip(names, counts)])
    print(f"Cluster {k}: {summary}")

# Plot clusters and centroids for cautious interpretation.
plt.figure(figsize=(7, 5))
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=70)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)
plt.xlabel("Vibration amplitude, mm/s")
plt.ylabel("Bearing temperature, degrees C")
plt.title("Clusters need engineering context before interpretation")
plt.grid(True, alpha=0.3)
plt.show()

```
