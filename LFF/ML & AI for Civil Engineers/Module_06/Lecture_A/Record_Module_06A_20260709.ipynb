## **1. Unsupervised Pattern Discovery**

### **1.1. No Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_01.jpg?v=1783620044" width="250">



>* Unlabeled data lack known answers
>* Find hidden patterns among engineering observations

>* No labels make correctness hard to judge
>* Unlabeled data still reveals useful patterns

>* Clusters depend on methods and chosen features
>* Interpret meaning with engineering judgment



```
#@title Python Code - No Labels

# Explore unlabeled engineering observations.
# Compare prediction with pattern discovery.
# Interpret clusters using engineering judgment.

# No extra installations are required.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create repeatable synthetic operating measurements.
rng = np.random.default_rng(7)
normal = rng.normal([22.0, 3.0], [1.0, 0.4], (18, 2))

# Create another unlabeled operating pattern.
high_load = rng.normal([31.0, 6.2], [1.2, 0.5], (18, 2))

# Combine observations without answer labels.
X = np.vstack([normal, high_load])
assert X.shape == (36, 2)

# Choose two starting centroids manually.
centroids = np.array([[21.0, 2.8], [32.0, 6.0]])
assert centroids.shape == (2, 2)

# Assign each point to nearest centroid.
distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
clusters = np.argmin(distances, axis=1)

# Update centroids from assigned observations.
updated = np.array([X[clusters == k].mean(axis=0) for k in range(2)])
assert np.isfinite(updated).all()

# Print concise teaching messages.
print("Supervised prediction needs known target labels.")
print("Here, observations have features but no answer key.")
print("K-means creates groups from similarity, not truth.")
print("Engineers must interpret clusters using context.")

# Plot unlabeled observations and updated centroids.
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", s=55)
plt.scatter(updated[:, 0], updated[:, 1], c="red", marker="X", s=180)
plt.xlabel("Temperature, degrees Celsius")
plt.ylabel("Vibration, millimeters per second")
plt.title("No Labels: Similarity-Based Groups")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Prediction Versus Discovery**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_02.jpg?v=1783620046" width="250">



>* Supervised models learn from known target labels
>* K-means discovers unlabeled engineering data patterns

>* Prediction uses errors; clustering uses pattern usefulness
>* Clusters suggest meanings, not confirmed answers

>* Interpret clusters with engineering context
>* Use clustering to guide further investigation



```
#@title Python Code - Prediction Versus Discovery

# Compare prediction with discovery.
# Use tiny engineering measurements.
# Interpret clusters cautiously.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create repeatable pump operating data.
rng = np.random.default_rng(7)
flow = np.array([42, 45, 47, 70, 73, 76, 50, 82])
vibration = np.array([0.18, 0.20, 0.22, 0.35, 0.38, 0.40, 0.75, 0.30])

# Combine features for clustering.
X = np.column_stack((flow, vibration))
assert X.shape == (8, 2)

# Choose two starting centroids manually.
centroids = X[[0, 6]].astype(float)
assert centroids.shape == (2, 2)

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([X[labels == k].mean(axis=0) for k in range(2)])

# Print concise teaching comparison.
print("Supervised prediction needs known target answers.")
print("Unsupervised discovery starts without target labels.")
print("K-means groups similar pump measurements only.")
print("Clusters suggest questions, not final engineering truth.")
print(f"Final centroids: {np.round(centroids, 2).tolist()}")

# Plot discovered groups and centroids.
plt.figure(figsize=(6, 4))
plt.scatter(flow, vibration, c=labels, s=90, cmap="viridis")
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)
plt.xlabel("Flow rate, gallons per minute")
plt.ylabel("Vibration level, inches per second")
plt.title("Discovery: clusters without known target labels")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.3. Engineering Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_01_03.jpg?v=1783620049" width="250">



>* Unsupervised patterns need engineering interpretation
>* Connect clusters to real system context

>* Clusters may reflect many hidden factors
>* Check patterns before making engineering decisions

>* Validate clusters before making engineering decisions
>* Use patterns to question, not conclude



```
#@title Python Code - Engineering Interpretation

# Interpret clusters using engineering context.
# Unsupervised groups need cautious explanations.
# This example uses fleet operations.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Set deterministic random behavior.
rng = np.random.default_rng(7)

# Create small fleet operating data.
light_load = rng.normal([8.0, 18.0], [0.7, 1.2], (10, 2))
heavy_load = rng.normal([15.0, 11.0], [0.9, 1.0], (10, 2))

# Combine observations into one dataset.
data = np.vstack([light_load, heavy_load])
assert data.shape == (20, 2)

# Choose two starting centroids manually.
centroids = np.array([[7.5, 17.0], [14.5, 12.0]])
assert centroids.shape == (2, 2)

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(data[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([data[labels == k].mean(axis=0) for k in range(2)])

# Summarize clusters without overclaiming causes.
for k in range(2):
    group = data[labels == k]
    mean_load = group[:, 0].mean()
    mean_mpg = group[:, 1].mean()
    print(f"Cluster {k}: {len(group)} vehicles, load {mean_load:.1f} tons, {mean_mpg:.1f} mpg")

# Print the engineering interpretation warning.
print("Clusters suggest patterns, not automatic causes.")
print("Check routes, payloads, weather, and maintenance records.")

# Plot clusters for cautious visual interpretation.
plt.figure(figsize=(6, 4))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", s=70)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=160)
plt.xlabel("Average payload, tons")
plt.ylabel("Fuel economy, miles per gallon")
plt.title("K-means pattern needing engineering interpretation")
plt.grid(True, alpha=0.3)
plt.show()

```

## **2. K Means Steps**

### **2.1. Initialize Centroids**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_01.jpg?v=1783620059" width="250">



>* Choose starting centroids before assigning data points
>* Initial locations can affect final clusters

>* Choose spread-out, sensible starting centroids
>* Use plots; updates refine imperfect choices

>* Use plots and engineering judgment
>* Avoid outliers as misleading starting centers



```
#@title Python Code - Initialize Centroids

# Demonstrate manual centroid initialization for k-means.
# Use bridge sensor readings as engineering data.
# Plot starting centroids before assignment steps.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create small deterministic sensor dataset.
strain_micro = np.array([120, 135, 150, 420, 455, 480, 760, 800, 835])
displacement_mm = np.array([2.1, 2.4, 2.8, 6.8, 7.2, 7.6, 12.0, 12.5, 13.1])

# Combine features into coordinate pairs.
points = np.column_stack((strain_micro, displacement_mm))
expected_shape = (9, 2)

# Validate the tiny dataset shape.
if points.shape != expected_shape:
    raise ValueError("Sensor data shape is not expected.")

# Choose three manual starting centroids.
centroids = np.array([[130, 2.3], [450, 7.2], [800, 12.6]])
centroid_labels = ["C1", "C2", "C3"]

# Validate centroid count and dimensions.
if centroids.shape != (3, 2):
    raise ValueError("Centroid array must contain three points.")

# Print a compact teaching summary.
print("Manual k-means initialization example")
print(f"Data points: {points.shape[0]}, features: {points.shape[1]}")
print("Centroids are starting assumptions, not final clusters.")
print("They should be distinct and spread across the plot.")

# Create one scatter plot.
plt.figure(figsize=(7, 5))
plt.scatter(strain_micro, displacement_mm, color="steelblue", label="Sensor readings")

# Add centroid markers to the plot.
plt.scatter(centroids[:, 0], centroids[:, 1], color="crimson", marker="X", s=180, label="Initial centroids")

# Label each centroid clearly.
for label, xy in zip(centroid_labels, centroids):
    plt.text(xy[0] + 10, xy[1] + 0.15, label)

# Add engineering context to axes.
plt.xlabel("Strain, microstrain")
plt.ylabel("Displacement, millimeters")
plt.title("Manual Initial Centroids for K-Means")

# Finish the single teaching plot.
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

```

### **2.2. Centroid Update**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_02.jpg?v=1783620057" width="250">



>* Move centroids to each cluster’s average position
>* Repeated updates improve rough initial centers

>* Recenter clusters using current assigned points
>* Repeat updates as memberships may change

>* Update centers from assigned cluster points
>* Check stability and interpret clusters cautiously



```
#@title Python Code - Centroid Update

# Manual centroid updates clarify k-means.
# Small route data keeps calculations visible.
# Engineering interpretation needs cautious checking.

import numpy as np
import matplotlib.pyplot as plt

# Store route distance and fuel use.
points = np.array([
    [2.0, 0.7], [2.5, 0.9],
    [8.0, 2.6], [9.0, 2.9],
    [3.0, 1.0], [8.5, 2.7]
])

# Start with two rough centroids.
centroids = np.array([
    [2.0, 0.7], [9.0, 2.9]
])

# Validate the tiny teaching dataset.
assert points.shape == (6, 2)
assert centroids.shape == (2, 2)

# Compute distances from points to centroids.
differences = points[:, None, :] - centroids[None, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))
assignments = np.argmin(distances, axis=1)

# Update each centroid using assigned points.
updated = centroids.copy()
for cluster_id in range(len(centroids)):
    members = points[assignments == cluster_id]
    if len(members) > 0:
        updated[cluster_id] = members.mean(axis=0)

# Summarize the manual update step.
print("Manual k-means centroid update")
print(f"Cluster assignments: {assignments.tolist()}")
print(f"Old centroids: {centroids.round(2).tolist()}")
print(f"New centroids: {updated.round(2).tolist()}")
print("Centroids are averages, not necessarily observed routes.")

# Plot points and centroid movement.
colors = np.array(["tab:blue", "tab:orange"])
plt.figure(figsize=(6, 4))
plt.scatter(points[:, 0], points[:, 1], c=colors[assignments], s=80)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=140, c="black")
plt.scatter(updated[:, 0], updated[:, 1], marker="*", s=180, c="red")

# Draw arrows from old to updated centroids.
for old, new in zip(centroids, updated):
    plt.arrow(old[0], old[1], new[0] - old[0], new[1] - old[1],
              length_includes_head=True, head_width=0.12, color="gray")

# Label the engineering features.
plt.xlabel("Route distance, miles")
plt.ylabel("Fuel use, gallons")
plt.title("Centroid update after one assignment step")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Centroid Update**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_02_03.jpg?v=1783620061" width="250">



>* Recalculate centroids using assigned points' feature averages
>* Centroids move toward each cluster's middle

>* Centroids recalibrate toward assigned points
>* Repeated updates make clusters more stable

>* Track assignments; centroids can shift easily
>* Stable clusters still need engineering judgment



```
#@title Python Code - Centroid Update

# Manual centroid updates summarize assigned engineering observations.
# This example uses tiny pump sensor data.
# We update centers without machine learning libraries.

import numpy as np
import matplotlib.pyplot as plt

# Store pressure and flow measurements for pumps.
points = np.array([
    [48.0, 102.0], [52.0, 98.0],
    [50.0, 105.0], [82.0, 158.0],

    [86.0, 162.0], [79.0, 155.0]
])

# Choose two starting centroids manually.
centroids = np.array([
    [45.0, 100.0], [90.0, 165.0]
])

# Confirm the tiny dataset has two features.
assert points.shape == (6, 2)
assert centroids.shape == (2, 2)

# Compute distances from every point to centroids.
differences = points[:, None, :] - centroids[None, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))

# Assign each point to its nearest centroid.
labels = np.argmin(distances, axis=1)
cluster_zero = points[labels == 0]
cluster_one = points[labels == 1]

# Update each centroid using assigned point averages.
updated_centroids = np.array([
    cluster_zero.mean(axis=0),
    cluster_one.mean(axis=0)
])

# Measure how far each centroid moved.
movement = np.sqrt(np.sum((updated_centroids - centroids) ** 2, axis=1))
print("Initial centroids:", centroids.tolist())
print("Updated centroids:", updated_centroids.round(2).tolist())
print("Cluster sizes:", [len(cluster_zero), len(cluster_one)])
print("Centroid movement:", movement.round(2).tolist())

# Plot points, old centroids, and updated centroids.
plt.figure(figsize=(6, 4))
plt.scatter(points[:, 0], points[:, 1], c=labels, s=80)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=140)
plt.scatter(updated_centroids[:, 0], updated_centroids[:, 1], marker="*", s=180)

# Label axes using engineering sensor units.
plt.xlabel("Pressure, psi")
plt.ylabel("Flow rate, gallons per minute")
plt.title("Manual K-Means Centroid Update")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.1. Cluster Context**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_03_01.jpg?v=1783620053" width="250">



>* Clusters show data patterns, not automatic meanings
>* Use engineering context before interpreting clusters

>* Use plots to inspect cluster separation
>* Check clusters against engineering context

>* Use engineering context before naming clusters
>* K-means shows patterns, not causes



```
#@title Python Code - Cluster Context

# Cluster context needs engineering judgment.
# K-means finds patterns, not causes.
# Plots help prevent unsupported labels.

import numpy as np
import matplotlib.pyplot as plt

# Create deterministic synthetic building data.
rng = np.random.default_rng(7)
small = rng.normal([80, 18], [8, 3], size=(18, 2))
large = rng.normal([260, 55], [18, 6], size=(18, 2))

# Combine observations into one dataset.
X = np.vstack([small, large])
assert X.shape == (36, 2)

# Manually choose two starting centroids.
centroids = np.array([[70, 20], [280, 50]], dtype=float)
assert centroids.shape == (2, 2)

# Run a few transparent k-means updates.
for step in range(4):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([X[labels == k].mean(axis=0) for k in range(2)])

# Summarize clusters without overclaiming causes.
counts = np.bincount(labels, minlength=2)
print("K-means grouped buildings by numerical similarity.")
print(f"Cluster sizes: {counts[0]} and {counts[1]} buildings.")
print("Here, separation is mostly building size and peak demand.")
print("Do not label clusters as usage types without context.")

# Plot clusters and centroids for interpretation.
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=70)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)
plt.xlabel("Monthly electricity use, thousand kWh")
plt.ylabel("Peak demand, kW")
plt.title("Cluster context: pattern first, explanation second")
plt.grid(True, alpha=0.3)
plt.show()

```

>* Clusters reveal possible operating patterns

```
#@title Python Code - Cluster Trends

# Cluster trends need cautious engineering interpretation.
# Synthetic sensor data keeps this example simple.
# Plots support hypotheses, not final conclusions.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make random values reproducible.
rng = np.random.default_rng(42)

# Create three operating-condition centers.
centers = np.array([
    [18.0, 2.0, 45.0],
    [30.0, 5.5, 70.0],
    [24.0, 3.5, 90.0],
])

# Generate compact engineering sensor clusters.
points = []
for center in centers:
    noise = rng.normal(0.0, [1.2, 0.5, 5.0], size=(18, 3))
    points.append(center + noise)

# Combine all observations into one array.
X = np.vstack(points)

# Validate the expected data shape.
assert X.shape == (54, 3)

# Choose manual starting centroids from observations.
centroids = X[[0, 20, 40]].copy()

# Run a few transparent k-means updates.
for step in range(6):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([X[labels == k].mean(axis=0) for k in range(3)])

# Summarize cluster trends without overclaiming.
print("K-means grouped 54 unlabeled operating records.")
print("Clusters suggest trends, not proven physical causes.")
for k in range(3):
    group = X[labels == k]
    print(f"Cluster {k}: n={len(group)}, temp={group[:,0].mean():.1f} C, vibration={group[:,1].mean():.1f} mm/s")

# Plot two variables while hiding occupancy.
plt.figure(figsize=(7, 5))
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=55)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=160)
plt.xlabel("Temperature, degrees C")
plt.ylabel("Vibration, mm per second")
plt.title("Cluster trends require engineering context")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Practical Engineering Judgment**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_A/image_03_03.jpg?v=1783620055" width="250">



>* Treat clusters as investigative hypotheses
>* Use engineering knowledge to explain patterns

>* Check data scaling and variable choices
>* Validate clusters with engineering knowledge

>* Use clusters to guide, not prove decisions
>* Revise interpretations with evidence and judgment



```
#@title Python Code - Practical Engineering Judgment

# Clusters need engineering judgment.
# Plots can mislead interpretation.
# Scaling changes k-means results.

# numpy and matplotlib are usually preinstalled.
import numpy as np
import matplotlib.pyplot as plt

# Create small building energy observations.
np.random.seed(7)
area = np.array([1200, 1350, 1500, 2100, 2300, 2500, 3200, 3400, 3600], dtype=float)
energy = np.array([42, 45, 47, 70, 73, 76, 95, 98, 101], dtype=float)

# Add occupancy as engineering context.
occupancy = np.array([2, 2, 3, 4, 4, 5, 4, 5, 5], dtype=float)
X_raw = np.column_stack([area, energy])

# Validate the tiny teaching dataset.
assert X_raw.shape == (9, 2)
assert occupancy.size == X_raw.shape[0]

# Define a simple k-means function.
def kmeans_manual(X, initial_centroids, steps=5):
    centroids = initial_centroids.astype(float).copy()
    for step in range(steps):
        distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
        labels = np.argmin(distances, axis=1)
        for group in range(centroids.shape[0]):
            if np.any(labels == group):
                centroids[group] = X[labels == group].mean(axis=0)
    return labels, centroids

# Run k-means on unscaled variables.
initial_raw = X_raw[[0, 3, 6]]
labels_raw, centers_raw = kmeans_manual(X_raw, initial_raw)

# Run k-means after unit scaling.
means = X_raw.mean(axis=0)
stds = X_raw.std(axis=0)
X_scaled = (X_raw - means) / stds

# Use matching initial scaled centroids.
initial_scaled = X_scaled[[0, 3, 6]]
labels_scaled, centers_scaled = kmeans_manual(X_scaled, initial_scaled)

# Print cautious interpretation messages.
print("Raw labels:", labels_raw.tolist())
print("Scaled labels:", labels_scaled.tolist())
print("Same data, different scaling can change cluster meaning.")
print("Check occupancy, logs, and physics before acting.")

# Plot scaled clusters with engineering context.
plt.figure(figsize=(7, 4.5))
scatter = plt.scatter(area, energy, c=labels_scaled, s=90, cmap="viridis")

# Annotate occupancy to encourage judgment.
for x_value, y_value, occ in zip(area, energy, occupancy):
    plt.text(x_value + 25, y_value, f"{int(occ)}p", fontsize=9)

# Label the engineering variables clearly.
plt.xlabel("Floor area, square feet")
plt.ylabel("Monthly energy use, MWh")
plt.title("K-means clusters need engineering interpretation")

# Add a compact legend and grid.
plt.colorbar(scatter, label="Scaled k-means cluster")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
