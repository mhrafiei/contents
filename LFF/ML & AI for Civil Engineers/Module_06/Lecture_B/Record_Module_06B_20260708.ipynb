### **1.1. Time Indexing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_01_01.jpg?v=1783564221" width="250">



>* Attach each measurement to a precise time
>* Use time-aware pandas or NumPy structures

>* Time indexes simplify reliable temporal data selection
>* They reveal spacing patterns and data gaps

>* Check time zones and timestamp meaning
>* Consistent indexing supports trustworthy analysis



```
#@title Python Code - Time Indexing

# Time indexes organize civil sensor observations.
# Pandas recognizes timestamps for reliable filtering.
# NumPy can store datetime values efficiently.

# Import small scientific computing libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create hourly timestamps for bridge monitoring.
times = pd.date_range(
    start="2026-03-01 00:00",
    periods=12,
    freq="h",
)

# Store vibration readings in metric units.
vibration_mm = np.array(
    [0.18, 0.20, 0.19, 0.23, 0.31, 0.42,
     0.39, 0.34, 0.28, 0.24, 0.21, 0.20]
)

# Build a DataFrame using timestamps as index.
bridge = pd.DataFrame(
    {"vibration_mm": vibration_mm},
    index=times,
)

# Name the index to document its meaning.
bridge.index.name = "timestamp"

# Validate the expected record count.
expected_count = 12
actual_count = len(bridge.index)
assert actual_count == expected_count

# Convert the index into NumPy datetime values.
numpy_times = bridge.index.to_numpy(
    dtype="datetime64[ns]"
)

# Select a meaningful morning inspection window.
morning = bridge.loc[
    "2026-03-01 06:00":"2026-03-01 09:00"
]

# Check whether readings are evenly spaced.
time_steps = np.diff(numpy_times)
regular_spacing = np.all(
    time_steps == time_steps[0]
)

# Print compact teaching results only.
print("First timestamp:", bridge.index[0])
print("Index type:", type(bridge.index).__name__)
print("Regular hourly spacing:", bool(regular_spacing))
print("Morning mean vibration:", round(morning["vibration_mm"].mean(), 3), "mm")

# Plot the indexed time series clearly.
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(bridge.index, bridge["vibration_mm"], marker="o")
ax.set_title("Bridge vibration readings with a time index")
ax.set_xlabel("Timestamp")

# Label the engineering measurement axis.
ax.set_ylabel("Vibration amplitude (mm)")
ax.grid(True, alpha=0.3)
fig.autofmt_xdate()
plt.tight_layout()

# Display exactly one plot.
plt.show()

```

### **1.2. Time Indexing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_01_02.jpg?v=1783564222" width="250">



>* Attach each measurement to its time
>* Use time to sort and compare data

>* Consistent timestamps enable practical time-series analysis
>* Time indexes align different civil data sources

>* Check timestamps for consistency and errors
>* Trustworthy time indexes support later analysis



```
#@title Python Code - Time Indexing

# Time indexes organize civil sensor observations.
# Pandas timestamps support filtering and alignment.
# This example uses bridge vibration readings.

# Import compact numerical and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create deterministic hourly timestamps for one day.
time_index = pd.date_range(
    start="2026-03-01 00:00",
    periods=24,
    freq="h",
)

# Build small synthetic acceleration measurements.
hours = np.arange(24)
accel_g = 0.015 + 0.006 * np.sin(hours / 24 * 2 * np.pi)
traffic_factor = np.where((hours >= 7) & (hours <= 18), 0.010, 0.002)

# Store measurements in a time-indexed DataFrame.
bridge_df = pd.DataFrame(
    {"acceleration_g": accel_g + traffic_factor},
    index=time_index,
)

# Name the index to document its meaning.
bridge_df.index.name = "timestamp"
assert bridge_df.index.is_monotonic_increasing
assert bridge_df.index.is_unique

# Select readings using clock time, not row numbers.
morning_df = bridge_df.between_time("07:00", "10:00")
evening_df = bridge_df.between_time("19:00", "22:00")

# Summarize the indexed data with concise output.
print("Rows and columns:", bridge_df.shape)
print("Index type:", type(bridge_df.index).__name__)
print("First timestamp:", bridge_df.index.min())
print("Last timestamp:", bridge_df.index.max())
print("Morning mean g:", round(morning_df["acceleration_g"].mean(), 4))
print("Evening mean g:", round(evening_df["acceleration_g"].mean(), 4))

# Plot the time-indexed vibration record.
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(bridge_df.index, bridge_df["acceleration_g"], marker="o")
ax.set_title("Bridge acceleration organized by time index")
ax.set_xlabel("Timestamp")
ax.set_ylabel("Acceleration, g")

# Improve readability for timestamp labels.
fig.autofmt_xdate()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Wide Time Tables

# Wide tables align many sensors by time.
# Each row represents one shared timestamp.
# Each column stores one civil measurement.

# Import common data analysis libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a small hourly timestamp index.
time_index = pd.date_range(
    start="2026-01-01 08:00",
    periods=8,
    freq="h"
)

# Build related bridge monitoring measurements.
wide_table = pd.DataFrame(
    {
        "deck_temp_C": [4.0, 4.5, 5.2, 6.1, 7.0, 7.4, 7.1, 6.5],
        "strain_microstrain": [102, 105, 111, 119, 128, 132, 126, 118],
        "wind_speed_mps": [3.1, 3.4, 4.0, 4.8, 5.2, 4.9, 4.1, 3.6],
        "traffic_count": [18, 25, 41, 58, 64, 52, 37, 24],
    },
    index=time_index
)

# Name the index to clarify its meaning.
wide_table.index.name = "timestamp"

# Validate the table has aligned rows.
expected_shape = (8, 4)
assert wide_table.shape == expected_shape

# Print compact structure information only.
print("Wide table shape:", wide_table.shape)
print("Index type:", type(wide_table.index).__name__)
print("Columns:", list(wide_table.columns))

# Show only the first three aligned timestamps.
print("First three rows:")
print(wide_table.head(3))

# Select one shared time window.
window = wide_table.loc["2026-01-01 10:00":"2026-01-01 13:00"]
print("Window mean traffic:", round(window["traffic_count"].mean(), 1))

# Plot two aligned signals on shared time.
ax = wide_table[["strain_microstrain", "traffic_count"]].plot(
    marker="o",
    figsize=(8, 4)
)

# Label the plot for interpretation.
ax.set_title("Wide Time Table: Aligned Bridge Measurements")
ax.set_xlabel("Timestamp")
ax.set_ylabel("Measurement value")
plt.tight_layout()
plt.show()

```

## **2. Temporal Feature Extraction**

### **2.1. Lag Feature Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_01.jpg?v=1783564209" width="250">



>* Past measurements help explain current conditions
>* Lag features capture time-series memory

>* Choose lags based on process and sampling.
>* Lags represent elapsed time and system response.

>* Use only past data to prevent leakage
>* Handle missing starts for reliable temporal patterns



```
#@title Python Code - Lag Feature Basics

# Create lag features for civil time series.
# Use pandas shifts without future leakage.
# Visualize current and previous traffic counts.

# Import compact numerical and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Build a small hourly traffic example.
hours = pd.date_range(
    "2026-01-05 06:00", periods=12, freq="h"
)

# Create deterministic traffic counts per hour.
traffic = np.array(
    [420, 510, 690, 760, 710, 640, 590, 620, 700, 810, 760, 650]
)

# Store measurements in a pandas table.
df = pd.DataFrame(
    {"traffic_vehicles_per_hour": traffic}, index=hours
)

# Add past values as lag features.
df["lag_1_hour"] = df["traffic_vehicles_per_hour"].shift(1)
df["lag_2_hours"] = df["traffic_vehicles_per_hour"].shift(2)

# Remove rows lacking earlier observations.
lagged = df.dropna().copy()

# Check that lagging preserved usable rows.
assert len(lagged) == len(df) - 2
assert lagged["lag_1_hour"].iloc[0] == df["traffic_vehicles_per_hour"].iloc[1]

# Print a compact teaching summary.
print("Lag features use only earlier observations.")
print("Original rows:", len(df), "usable lagged rows:", len(lagged))
print("First usable time:", lagged.index[0].strftime("%H:%M"))
print("Current traffic:", int(lagged.iloc[0, 0]), "vehicles/hour")
print("One-hour lag:", int(lagged.iloc[0, 1]), "vehicles/hour")

# Plot current traffic and one-hour lag.
plt.figure(figsize=(8, 4))
plt.plot(lagged.index, lagged["traffic_vehicles_per_hour"], marker="o", label="current")
plt.plot(lagged.index, lagged["lag_1_hour"], marker="s", label="lag 1 hour")
plt.title("Traffic Time Series With Lag Feature")
plt.xlabel("Time of day")
plt.ylabel("Vehicles per hour")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **2.2. Rolling Averages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_02.jpg?v=1783564211" width="250">



>* Smooth noisy measurements using recent observations
>* Reveal trends and sustained system changes

>* Short windows react quickly but stay noisy
>* Long windows smooth trends but delay changes

>* Use only past data for forecasting
>* Combine rolling averages with other time features



```
#@title Python Code - Rolling Averages

# Rolling averages smooth noisy engineering sensor readings.
# Window length controls responsiveness and smoothness.
# Past-only features prevent future information leakage.

# Import numerical, tabular, and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic hourly pressure signal.
np.random.seed(7)
hours = pd.date_range("2026-01-01", periods=36, freq="h")
base_pressure = 62 + 3 * np.sin(np.arange(36) / 3)
noise = np.random.normal(0, 1.2, size=36)

# Add a sustained pressure drop after hour twenty.
pressure_psi = base_pressure + noise
pressure_psi[20:] = pressure_psi[20:] - 5
sensor_data = pd.DataFrame({"pressure_psi": pressure_psi}, index=hours)

# Compute past-only rolling averages for two windows.
sensor_data["roll_3h"] = sensor_data["pressure_psi"].rolling(3).mean()
sensor_data["roll_8h"] = sensor_data["pressure_psi"].rolling(8).mean()
valid_rows = sensor_data.dropna()

# Validate enough rows exist for plotting.
assert len(valid_rows) >= 20, "Need enough rolling-average rows."
print("Rows after rolling windows:", len(valid_rows))
print("Latest pressure psi:", round(sensor_data["pressure_psi"].iloc[-1], 2))
print("Latest 3-hour average:", round(sensor_data["roll_3h"].iloc[-1], 2))
print("Latest 8-hour average:", round(sensor_data["roll_8h"].iloc[-1], 2))

# Plot raw pressure and rolling averages.
plt.figure(figsize=(9, 4))
plt.plot(sensor_data.index, sensor_data["pressure_psi"], label="Raw pressure")
plt.plot(sensor_data.index, sensor_data["roll_3h"], label="3-hour average")
plt.plot(sensor_data.index, sensor_data["roll_8h"], label="8-hour average")

# Label the engineering time-series plot.
plt.title("Water Pressure Rolling Averages")
plt.xlabel("Time")
plt.ylabel("Pressure, psi")
plt.legend()

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```

### **2.3. Rolling Averages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_03.jpg?v=1783564213" width="250">



>* Smooth noisy measurements using recent history
>* Reveal meaningful trends while preserving time order

>* Window length balances speed and smoothness
>* Rolling averages support forecasting and anomaly detection

>* Use only available past and current data
>* Combine features to capture system behavior



```
#@title Python Code - Rolling Averages

# Rolling averages smooth noisy civil time series.
# Pandas creates features from recent observations.
# This example uses bridge traffic speeds.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the random example reproducible.
rng = np.random.default_rng(seed=42)
periods = 36
hours = pd.date_range("2026-01-01 06:00", periods=periods, freq="10min")

# Simulate bridge speeds in miles per hour.
trend = np.linspace(58, 38, periods)
noise = rng.normal(loc=0, scale=3, size=periods)
speeds = np.clip(trend + noise, 15, 70)

# Store the time series in pandas.
data = pd.DataFrame({"speed_mph": speeds}, index=hours)
assert len(data) == periods
assert data["speed_mph"].notna().all()

# Compute rolling averages using past observations.
data["roll_3"] = data["speed_mph"].rolling(window=3).mean()
data["roll_6"] = data["speed_mph"].rolling(window=6).mean()
valid_rows = data.dropna().copy()

# Print a compact feature preview.
print("Rolling average feature preview:")
print(valid_rows[["speed_mph", "roll_3", "roll_6"]].head(5).round(1))
print("Short windows react faster; long windows smooth more.")

# Plot raw speeds and rolling averages.
plt.figure(figsize=(9, 4))
plt.plot(data.index, data["speed_mph"], marker="o", label="Observed speed")
plt.plot(data.index, data["roll_3"], linewidth=2, label="3-reading average")

# Add the smoother six-reading average.
plt.plot(data.index, data["roll_6"], linewidth=2, label="6-reading average")
plt.title("Bridge Traffic Speed Rolling Averages")
plt.xlabel("Time")
plt.ylabel("Speed, miles per hour")

# Finish the single teaching plot.
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

## **3. Forecast Baselines**

### **3.1. Naive Forecasts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_01.jpg?v=1783564215" width="250">



>* Predict next value from recent observations
>* Set a simple benchmark for better models

>* Match naive forecasts to data patterns
>* Use simple baselines for model comparison

>* Compare forecasts with actual values
>* Use errors as interpretable diagnostic benchmarks



```
#@title Python Code - Naive Forecasts

# Naive forecasts reuse recent observed values.
# Civil time series need simple baselines.
# Errors reveal missed temporal behavior.

# Import common numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a small hourly traffic dataset.
np.random.seed(7)
hours = pd.date_range("2024-01-01", periods=48, freq="h")
base_flow = 900 + 180 * np.sin(np.arange(48) * np.pi / 12)
noise = np.random.normal(0, 35, size=48)

# Store observations in a pandas DataFrame.
data = pd.DataFrame({"vehicles_per_hour": base_flow + noise}, index=hours)
data["naive_forecast"] = data["vehicles_per_hour"].shift(1)
data["error"] = data["vehicles_per_hour"] - data["naive_forecast"]

# Drop the first row without forecast.
valid = data.dropna().copy()
assert len(valid) == 47
mae = valid["error"].abs().mean()
rmse = np.sqrt((valid["error"] ** 2).mean())

# Print compact baseline evaluation results.
print("Naive forecast: next hour equals previous hour.")
print(f"Mean absolute error: {mae:.1f} vehicles/hour")
print(f"Root mean squared error: {rmse:.1f} vehicles/hour")
print("Large errors often occur during rapid traffic changes.")

# Plot actual values, forecasts, and errors.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(valid.index, valid["vehicles_per_hour"], label="Actual", linewidth=2)
ax.plot(valid.index, valid["naive_forecast"], label="Naive forecast", linestyle="--")
ax.bar(valid.index, valid["error"], label="Error", alpha=0.25)

# Add readable labels for interpretation.
ax.set_title("Naive Forecast Baseline for Hourly Traffic Flow")
ax.set_ylabel("Vehicles per hour")
ax.set_xlabel("Time")
ax.legend(loc="upper right")

# Improve spacing for Google Colab display.
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```

### **3.2. Residual Error Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_02.jpg?v=1783564219" width="250">



>* Residuals show forecast errors over time
>* Random-looking errors suggest adequate baselines

>* Residual patterns reveal baseline forecasting failures
>* Structured errors suggest better models can improve

>* Investigate unusual residuals before removing them
>* Compare baselines to guide model improvements



```
#@title Python Code - Residual Error Plots

# Residual plots reveal baseline forecasting mistakes.
# Civil time series often contain patterns.
# This example uses synthetic traffic counts.

# Import numerical, table, and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the example deterministic and small.
rng = np.random.default_rng(42)
periods = 56

# Create daily dates for eight weeks.
dates = pd.date_range(
    start="2024-01-01", periods=periods, freq="D"
)

# Build trend, weekly cycle, and noise.
day_index = np.arange(periods)
trend = 8 * day_index
weekly = 180 * np.sin(2 * np.pi * day_index / 7)
noise = rng.normal(loc=0, scale=45, size=periods)

# Combine components into traffic counts.
traffic = 2200 + trend + weekly + noise
traffic = np.maximum(traffic, 0).round()

# Store the time series in pandas.
data = pd.DataFrame({"traffic_count": traffic}, index=dates)
assert len(data) == periods

# Create a naive persistence forecast.
data["forecast"] = data["traffic_count"].shift(1)
data["residual"] = data["traffic_count"] - data["forecast"]

# Drop the first row with missing forecast.
results = data.dropna().copy()
assert results["residual"].notna().all()

# Summarize residual behavior with few lines.
mae = results["residual"].abs().mean()
bias = results["residual"].mean()
print(f"Rows evaluated: {len(results)}")
print(f"Naive baseline MAE: {mae:.1f} vehicles/day")
print(f"Mean residual bias: {bias:.1f} vehicles/day")

# Plot residuals over time for diagnosis.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(results.index, results["residual"], marker="o", linewidth=1.5)
ax.axhline(0, color="black", linewidth=1)

# Label the residual plot clearly.
ax.set_title("Residual Error Plot for Naive Traffic Forecast")
ax.set_xlabel("Date")
ax.set_ylabel("Actual minus forecast, vehicles/day")
ax.grid(True, alpha=0.3)

# Rotate dates and display the single plot.
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```

### **3.3. Error Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_03.jpg?v=1783564217" width="250">



>* Plot errors to reveal hidden forecast failures
>* Identify when mistakes follow meaningful patterns

>* Plot errors against time to reveal patterns
>* Identify baseline weaknesses during changing conditions

>* Show forecast limits for practical decisions
>* Use visible errors to guide engineering judgment



```
#@title Python Code - Error Visualization

# Visualize baseline forecast errors over time.
# Civil time series often hide patterns.
# Residual plots reveal when baselines fail.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the synthetic example deterministic.
np.random.seed(7)
periods = 72
hours = pd.date_range("2026-01-01", periods=periods, freq="h")

# Create hourly bridge strain in microstrain.
time_index = np.arange(periods)
daily_cycle = 18 * np.sin(2 * np.pi * time_index / 24)
trend = 0.12 * time_index

# Add a short storm-related disturbance.
storm_effect = np.zeros(periods)
storm_effect[42:52] = np.linspace(0, 28, 10)
noise = np.random.normal(0, 3, periods)

# Store observations in a pandas DataFrame.
strain = 210 + daily_cycle + trend + storm_effect + noise
data = pd.DataFrame({"strain_microstrain": strain}, index=hours)

# Build a naive one-step baseline forecast.
data["forecast"] = data["strain_microstrain"].shift(1)
data["error"] = data["strain_microstrain"] - data["forecast"]

# Remove the first row without a forecast.
valid = data.dropna().copy()
assert len(valid) == periods - 1

# Summarize errors without printing large tables.
mae = valid["error"].abs().mean()
bias = valid["error"].mean()
max_error_time = valid["error"].abs().idxmax()

# Print a compact diagnostic summary.
print(f"Naive baseline MAE: {mae:.2f} microstrain")
print(f"Average signed error: {bias:.2f} microstrain")
print(f"Largest error time: {max_error_time:%Y-%m-%d %H:%M}")

# Plot residuals to reveal timing patterns.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(valid.index, valid["error"], marker="o", linewidth=1.2)
ax.axhline(0, color="black", linewidth=1)

# Highlight the storm period where errors grow.
ax.axvspan(hours[42], hours[52], color="orange", alpha=0.2)
ax.set_title("Naive Forecast Errors for Bridge Strain")
ax.set_ylabel("Observed minus forecast, microstrain")
ax.set_xlabel("Time")

# Improve readability for online lecture viewing.
ax.grid(True, alpha=0.3)
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```
