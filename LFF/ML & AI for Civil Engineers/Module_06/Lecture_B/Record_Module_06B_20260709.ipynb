```
#@title Python Code - Time Indexing

# Time indexes organize engineering measurements.
# Pandas stores timestamps as meaningful indexes.
# Small examples reveal temporal structure.

# Import common time-series analysis libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create hourly timestamps for bridge monitoring.
times = pd.date_range(
    start="2026-03-01 06:00",
    periods=8,
    freq="h"
)

# Store strain readings with timestamp index.
strain = np.array([118, 121, 125, 130, 128, 124, 122, 119])
bridge = pd.DataFrame({"strain_microstrain": strain}, index=times)

# Name the index to document meaning.
bridge.index.name = "measurement_time"

# Select a rush-hour time window.
rush_hour = bridge.loc["2026-03-01 08:00":"2026-03-01 10:00"]

# Check whether timestamps are evenly spaced.
time_steps = bridge.index.to_series().diff().dropna()
regular_spacing = time_steps.nunique() == 1

# Print compact teaching results.
print("First timestamp:", bridge.index[0])
print("Last timestamp:", bridge.index[-1])
print("Regular hourly spacing:", regular_spacing)
print("Rush-hour mean strain:", round(rush_hour["strain_microstrain"].mean(), 1))
print("Rows and columns:", bridge.shape)

# Plot the indexed strain series.
ax = bridge.plot(
    y="strain_microstrain",
    marker="o",
    legend=False,
    figsize=(7, 3)
)

# Label the plot for interpretation.
ax.set_title("Bridge Strain Readings With Time Index")
ax.set_xlabel("Measurement time")
ax.set_ylabel("Strain, microstrain")
plt.tight_layout()
plt.show()

```

### **1.2. Time Indexing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_01_02.jpg?v=1783621294" width="250">



>* Link each measurement to its observation time
>* Use pandas indexes with NumPy values

>* Time indexes reveal spacing between observations

>* Find missing, duplicate, or irregular timestamps

```
#@title Python Code - Time Indexing

# Time indexes organize civil sensor observations.
# Pandas stores timestamps beside measured values.
# NumPy stores compact numeric measurement arrays.

# Import numerical and time-series libraries.
import numpy as np
import pandas as pd

# Create six hourly bridge strain timestamps.
times = pd.date_range(
    start="2026-03-01 00:00",
    periods=6,
    freq="h"
)

# Store strain readings in a NumPy array.
strain_micro = np.array(
    [118, 121, 125, 130, 128, 126],
    dtype=float
)

# Validate matching timestamp and value counts.
assert len(times) == strain_micro.size

# Build a pandas Series with time indexing.
strain_series = pd.Series(
    data=strain_micro,
    index=times,
    name="bridge_strain_microstrain"
)

# Select readings during the morning window.
morning = strain_series.loc[
    "2026-03-01 02:00":"2026-03-01 04:00"
]

# Create an irregular inspection log.
inspection_times = pd.to_datetime(
    ["2026-03-01 00:30", "2026-03-01 03:15", "2026-03-01 05:45"]
)

# Store pavement rut depths in inches.
rut_inches = np.array(
    [0.18, 0.22, 0.21],
    dtype=float
)

# Build a time-indexed inspection DataFrame.
inspection = pd.DataFrame(
    {"rut_depth_inches": rut_inches},
    index=inspection_times
)

# Check spacing between bridge sensor readings.
spacing = strain_series.index.to_series().diff().dropna()

# Print compact teaching outputs only.
print("Bridge strain index type:", type(strain_series.index).__name__)
print("First timestamp:", strain_series.index[0])
print("Morning strain values:", morning.to_list())
print("Regular hourly spacing:", spacing.nunique() == 1)
print("Inspection timestamps irregular:", inspection.index.to_series().diff().nunique() > 1)
print("Inspection rows stored:", len(inspection))

```

## **2. Temporal Feature Extraction**

### **2.1. Lag Feature Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_01.jpg?v=1783621307" width="250">



>* Past measurements help predict current conditions
>* Lag features reveal delayed engineering patterns

>* Lags capture persistence, delays, and repeated patterns
>* Lag meaning depends on sampling and system behavior

>* Use only past information for predictions
>* Handle missing and redundant lag values



```
#@title Python Code - Lag Feature Basics

# Lag features reveal recent system memory.
# Civil sensors often depend on history.
# This example uses bridge traffic counts.

# Import compact numerical and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a small deterministic hourly dataset.
hours = pd.date_range(
    start="2026-01-01 06:00",
    periods=24,
    freq="h"
)

# Build traffic with daily shape.
base_counts = np.array(
    [120, 180, 260, 340, 410, 390,
     300, 240, 220, 260, 310, 360,
     420, 460, 430, 380, 330, 290,
     250, 210, 180, 160, 140, 130]
)

# Store observations in a pandas table.
data = pd.DataFrame(
    {"traffic_count": base_counts},
    index=hours
)

# Add past values as lag features.
data["lag_1_hour"] = data["traffic_count"].shift(1)
data["lag_3_hours"] = data["traffic_count"].shift(3)
data["lag_6_hours"] = data["traffic_count"].shift(6)

# Remove rows lacking earlier observations.
feature_table = data.dropna().copy()
assert len(feature_table) > 0, "Lag table unexpectedly empty."

# Show only a compact teaching preview.
print("Lag features use only earlier traffic counts.")
print(feature_table.head(4).round(0).to_string())

# Compare current traffic with one-hour lag.
correlation_value = feature_table[
    "traffic_count"
].corr(feature_table["lag_1_hour"])

# Print one concise relationship summary.
print(f"Correlation with one-hour lag: {correlation_value:.2f}")
print("Missing early rows were removed after shifting.")

# Plot current and lagged traffic together.
plt.figure(figsize=(8, 4))
plt.plot(data.index, data["traffic_count"], marker="o", label="current")
plt.plot(data.index, data["lag_1_hour"], marker="s", label="lag 1 hour")

# Label the plot for interpretation.
plt.title("Bridge traffic and one-hour lag feature")
plt.xlabel("Time")
plt.ylabel("Vehicles per hour")
plt.legend()

# Keep the time labels readable.
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

### **2.2. Rolling Averages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_02.jpg?v=1783621305" width="250">



>* Smooth noisy measurements using recent observation windows
>* Reveal trends, shifts, and unusual behavior

>* Window size controls smoothness and responsiveness
>* Match windows to system and timescale

>* Summarize recent history for models and diagnostics
>* Avoid future data in rolling features



```
#@title Python Code - Rolling Averages

# Rolling averages smooth noisy engineering measurements.
# Window size controls responsiveness and smoothness.
# Pandas makes temporal feature extraction simple.

# Import numerical, tabular, and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic hourly sensor time series.
np.random.seed(7)
hours = pd.date_range("2026-01-01", periods=48, freq="h")
trend = np.linspace(2.0, 3.2, len(hours))

# Add daily pattern and small sensor noise.
daily_cycle = 0.35 * np.sin(np.arange(48) * np.pi / 12)
noise = np.random.normal(0.0, 0.12, len(hours))
water_level_m = trend + daily_cycle + noise

# Store measurements in a pandas DataFrame.
df = pd.DataFrame({"water_level_m": water_level_m}, index=hours)
assert df.shape[0] >= 6, "Need enough observations."

# Compute short and longer rolling averages.
df["roll_3h_m"] = df["water_level_m"].rolling(window=3).mean()
df["roll_12h_m"] = df["water_level_m"].rolling(window=12).mean()

# Show only a compact teaching summary.
print("Rows:", len(df), "hourly water-level readings")
print("Raw standard deviation:", round(df["water_level_m"].std(), 3))
print("3-hour rolling standard deviation:", round(df["roll_3h_m"].std(), 3))
print("12-hour rolling standard deviation:", round(df["roll_12h_m"].std(), 3))
print("Last raw level:", round(df["water_level_m"].iloc[-1], 3), "m")
print("Last 12-hour average:", round(df["roll_12h_m"].iloc[-1], 3), "m")

# Plot raw readings and rolling averages together.
plt.figure(figsize=(9, 4))
plt.plot(df.index, df["water_level_m"], label="Raw level", alpha=0.55)
plt.plot(df.index, df["roll_3h_m"], label="3-hour average", linewidth=2)
plt.plot(df.index, df["roll_12h_m"], label="12-hour average", linewidth=2)

# Label the engineering context clearly.
plt.title("Rolling Averages for Water-Level Monitoring")
plt.xlabel("Time")
plt.ylabel("Water level, meters")
plt.legend()

# Improve readability and display the single plot.
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **2.3. Rolling Averages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_02_03.jpg?v=1783621309" width="250">



>* Smooth noisy measurements using recent observations
>* Reveal broader trends in civil systems

>* Window length balances speed and smoothness
>* Rolling averages summarize recent time history

>* Separate short spikes from lasting trends
>* Avoid future data in rolling features



```
#@title Python Code - Rolling Averages

# Rolling averages smooth noisy civil sensor readings.
# Recent windows summarize short term behavior.
# Careful shifting prevents future data leakage.

# Import common numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a reproducible hourly bridge strain series.
rng = np.random.default_rng(42)
time_index = pd.date_range("2026-01-01", periods=48, freq="h")
trend = np.linspace(100, 112, len(time_index))
noise = rng.normal(0, 2.5, len(time_index))

# Add a short traffic related spike.
strain = trend + noise
strain[24:28] = strain[24:28] + 10

# Store measurements in a tidy DataFrame.
data = pd.DataFrame({"strain_microstrain": strain}, index=time_index)
assert len(data) == 48

# Compute rolling averages using past observations only.
data["roll_3h"] = data["strain_microstrain"].rolling(3).mean()
data["roll_12h"] = data["strain_microstrain"].rolling(12).mean()

# Shift features for prediction without future leakage.
data["safe_roll_3h"] = data["roll_3h"].shift(1)
data["safe_roll_12h"] = data["roll_12h"].shift(1)

# Print a compact teaching summary.
print("Rows:", len(data), "hourly bridge strain readings")
print("Short window reacts faster; long window smooths more.")
print("Safe rolling features are shifted one hour backward.")
print(data[["strain_microstrain", "safe_roll_3h", "safe_roll_12h"]].tail(5).round(2))

# Plot raw readings and rolling averages together.
plt.figure(figsize=(9, 4))
plt.plot(data.index, data["strain_microstrain"], label="raw strain", alpha=0.55)
plt.plot(data.index, data["safe_roll_3h"], label="safe 3-hour average")
plt.plot(data.index, data["safe_roll_12h"], label="safe 12-hour average")

# Label the civil engineering time series plot.
plt.title("Rolling Averages for Bridge Strain Monitoring")
plt.xlabel("Time")
plt.ylabel("Strain, microstrain")
plt.legend()

# Display the single required visualization.
plt.tight_layout()
plt.show()

```

## **3. Forecast Baselines**

### **3.1. Naive Forecasts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_01.jpg?v=1783621299" width="250">



>* Predict next value from recent observations
>* Use baselines to judge complex models

>* Recent values often predict short-term behavior
>* Seasonal baselines handle repeating time patterns

>* Naive baselines set prediction quality standards
>* Failures reveal when advanced models are needed



```
#@title Python Code - Naive Forecasts

# Naive forecasts reuse the latest observed value.
# Civil sensors often show short term persistence.
# Baselines reveal whether complex models add value.

# Import compact numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic hourly traffic pattern.
rng = np.random.default_rng(42)
periods = 48
hours = pd.date_range("2026-01-01", periods=periods, freq="h")

# Combine daily cycles with small random noise.
cycle = 120 + 35 * np.sin(np.arange(periods) * np.pi / 12)
noise = rng.normal(0, 8, periods)
traffic = np.round(cycle + noise).astype(int)

# Store observations in a pandas DataFrame.
data = pd.DataFrame({"vehicles": traffic}, index=hours)
data["naive_forecast"] = data["vehicles"].shift(1)
data["error"] = data["vehicles"] - data["naive_forecast"]

# Drop the first row without a forecast.
valid = data.dropna().copy()
assert len(valid) == periods - 1
assert valid["naive_forecast"].notna().all()

# Compute simple baseline error metrics.
mae = valid["error"].abs().mean()
rmse = np.sqrt((valid["error"] ** 2).mean())
last_error = valid["error"].iloc[-1]

# Print a compact teaching summary.
print("Naive forecast: next hour equals previous hour.")
print(f"Rows evaluated: {len(valid)} hourly traffic counts.")
print(f"Mean absolute error: {mae:.1f} vehicles.")
print(f"Root mean squared error: {rmse:.1f} vehicles.")
print(f"Last forecast error: {last_error:.1f} vehicles.")

# Plot actual values, forecasts, and errors.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(valid.index, valid["vehicles"], label="Actual", linewidth=2)
ax.plot(valid.index, valid["naive_forecast"], label="Naive forecast", linestyle="--")

# Add error bars between actual and forecast.
ax.vlines(valid.index, valid["naive_forecast"], valid["vehicles"], color="gray", alpha=0.35)
ax.set_title("Naive Forecast Baseline for Hourly Traffic")
ax.set_ylabel("Vehicles per hour")
ax.legend()

# Improve readability for time labels.
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```

### **3.2. Residual Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_02.jpg?v=1783621301" width="250">



>* Residuals compare observations with baseline forecasts
>* Plots reveal when and why forecasts fail

>* Plot residuals over time to reveal patterns
>* Structured errors expose missed real-world influences

>* Separate random errors from recurring problems
>* Use patterns to guide model or system checks



```
#@title Python Code - Residual Visualization

# Residual plots reveal baseline forecast mistakes.
# Civil time series often contain rhythms.
# Simple visuals support engineering diagnosis.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic hourly traffic example.
np.random.seed(7)
periods = 72
hours = pd.date_range(
    "2026-01-01",
    periods=periods,
    freq="h"
)

# Build daily traffic pattern with noise.
hour_index = np.arange(periods)
daily_wave = 180 * np.sin(2 * np.pi * hour_index / 24)
trend = 2.5 * hour_index
noise = np.random.normal(0, 35, periods)

# Store observed traffic counts in pandas.
traffic = 900 + daily_wave + trend + noise
data = pd.DataFrame({"vehicles_per_hour": traffic}, index=hours)

# Create a naive one hour baseline forecast.
data["forecast"] = data["vehicles_per_hour"].shift(1)
data["residual"] = data["vehicles_per_hour"] - data["forecast"]

# Remove the first row without a forecast.
valid = data.dropna().copy()
assert len(valid) == periods - 1

# Summarize residual size and direction.
mae = valid["residual"].abs().mean()
bias = valid["residual"].mean()
rmse = np.sqrt((valid["residual"] ** 2).mean())

# Print a compact diagnostic summary.
print(f"Rows evaluated: {len(valid)} hourly observations")
print(f"Mean absolute error: {mae:.1f} vehicles per hour")
print(f"Average residual bias: {bias:.1f} vehicles per hour")
print(f"Root mean squared error: {rmse:.1f} vehicles per hour")

# Plot residuals against time for diagnosis.
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(valid.index, valid["residual"], marker="o", linewidth=1)
ax.axhline(0, color="black", linewidth=1)

# Add labels that explain the residual plot.
ax.set_title("Residuals from a Naive Traffic Forecast")
ax.set_xlabel("Time")
ax.set_ylabel("Observed minus forecast")
ax.grid(True, alpha=0.3)

# Display the single teaching plot.
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```

### **3.3. Error Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_06/Lecture_B/image_03_03.jpg?v=1783621303" width="250">



>* Plot errors to reveal hidden forecast patterns
>* Timing matters during critical engineering conditions

>* Compare forecasts and observations over time
>* Spot bias, lag, and missing patterns

>* Error plots reveal baseline reliability under stress
>* Patterns explain mistakes and guide model benchmarks



```
#@title Python Code - Error Visualization

# Visualize baseline forecast errors clearly.
# Use synthetic civil time-series data.
# Compare predictions with observed values.

# Import common analysis and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the example deterministic and small.
rng = np.random.default_rng(42)
periods = 60

# Create daily dates for bridge strain monitoring.
dates = pd.date_range(
    start="2025-01-01", periods=periods, freq="D"
)

# Simulate observed strain in microstrain units.
trend = np.linspace(100, 112, periods)
cycle = 6 * np.sin(np.arange(periods) * 2 * np.pi / 7)

# Add small noise and two event spikes.
noise = rng.normal(0, 1.8, periods)
observed = trend + cycle + noise

# Add unusual loading events to observed strain.
observed[24] += 18
observed[43] -= 14

# Store the time series in pandas.
data = pd.DataFrame({"observed_microstrain": observed}, index=dates)

# Build a naive persistence baseline forecast.
data["forecast_microstrain"] = data["observed_microstrain"].shift(1)
data = data.dropna().copy()

# Compute signed and absolute forecast errors.
data["error"] = data["observed_microstrain"] - data["forecast_microstrain"]
data["absolute_error"] = data["error"].abs()

# Validate that aligned series have matching lengths.
assert len(data["observed_microstrain"]) == len(data["forecast_microstrain"])

# Summarize errors without printing full tables.
mae = data["absolute_error"].mean()
bias = data["error"].mean()

# Print concise interpretation for learners.
print(f"Naive baseline MAE: {mae:.2f} microstrain")
print(f"Average signed error: {bias:.2f} microstrain")
print("Positive error means the baseline underpredicted strain.")

# Create one figure with forecast and error views.
fig, axes = plt.subplots(2, 1, figsize=(10, 7), sharex=True)

# Plot observed and predicted strain together.
axes[0].plot(data.index, data["observed_microstrain"], label="Observed")
axes[0].plot(data.index, data["forecast_microstrain"], label="Naive forecast")
axes[0].set_ylabel("Microstrain")

# Add legend and title for comparison.
axes[0].legend(loc="upper left")
axes[0].set_title("Bridge strain baseline forecast")

# Plot signed errors around a zero reference.
axes[1].bar(data.index, data["error"], color="steelblue")
axes[1].axhline(0, color="black", linewidth=1)
axes[1].set_ylabel("Error")

# Label the time axis and error meaning.
axes[1].set_xlabel("Date")
axes[1].set_title("Forecast error over time")

# Improve spacing and display the single plot.
fig.autofmt_xdate()
plt.tight_layout()
plt.show()

```
