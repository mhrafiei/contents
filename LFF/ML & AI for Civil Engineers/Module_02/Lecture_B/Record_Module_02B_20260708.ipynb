## **1. Data Cleaning Basics**

### **1.1. Missing Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1783553898" width="250">



>* Identify missing data and possible forms
>* Understand causes before choosing treatment

>* Choose removal or imputation carefully
>* Compare results before and after cleaning

>* Check missing data patterns for bias
>* Document cleaning choices for transparent decisions



```
#@title Python Code - Missing Values

# Missing values can hide engineering data problems.
# Small datasets need careful cleaning decisions.
# This example compares simple cleaning choices.

import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny civil engineering inspection dataset.
data = {
    "site": ["A", "B", "C", "D", "E", "F"],
    "rainfall_mm": [12.0, None, 0.0, 28.0, None, 18.0],

    "traffic_count": [420, 390, None, 610, 580, None],
    "condition_rating": [4, None, 3, 2, None, 5],
}

# Convert the dictionary into a dataframe.
df = pd.DataFrame(data)

# Count missing values in each column.
missing_counts = df.isna().sum()

# Treat impossible zero rainfall as missing here.
df["rainfall_mm"] = df["rainfall_mm"].replace(0.0, pd.NA)

# Fill rainfall using the column median.
rain_median = df["rainfall_mm"].median()
df["rainfall_filled"] = df["rainfall_mm"].fillna(rain_median)

# Fill traffic using the column mean.
traffic_mean = df["traffic_count"].mean()
df["traffic_filled"] = df["traffic_count"].fillna(traffic_mean)

# Keep rows with known condition ratings.
known_condition = df.dropna(subset=["condition_rating"])

# Print compact summaries for learners.
print("Original missing counts:", missing_counts.to_dict())
print("Rainfall median used:", round(rain_median, 1), "mm")
print("Traffic mean used:", round(traffic_mean, 1), "vehicles")
print("Rows kept for condition analysis:", len(known_condition))
print("Cleaned rainfall values:", df["rainfall_filled"].tolist())

# Plot original and filled rainfall values.
plt.figure(figsize=(7, 4))
plt.plot(df["site"], df["rainfall_filled"], marker="o", label="filled")
plt.scatter(df["site"], df["rainfall_mm"], color="red", label="original")

# Label the plot clearly.
plt.title("Missing Rainfall Values Before and After Filling")
plt.xlabel("Inspection site")
plt.ylabel("Rainfall during inspection, mm")
plt.legend()

# Display the single teaching plot.
plt.tight_layout()
plt.show()

```

### **1.2. Duplicate Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1783553900" width="250">



>* Repeated records can arise from collection errors
>* Duplicates can bias engineering analysis results

>* Distinguish exact from partial duplicates
>* Use key fields to judge uniqueness

>* Handle duplicates based on their true meaning
>* Document decisions for transparent cleaned datasets



```
#@title Python Code - Duplicate Data

# Duplicate records can bias engineering summaries.
# Key columns define unique observations.
# Cleaning choices should be documented.

# Import pandas for small tabular cleaning.
import pandas as pd

# Create a tiny pavement inspection dataset.
records = [
    {"segment": "A12", "date": "2026-01-05", "crack_pct": 12, "rut_in": 0.18},
    {"segment": "A12", "date": "2026-01-05", "crack_pct": 12, "rut_in": 0.18},

    {"segment": "B07", "date": "2026-01-05", "crack_pct": 28, "rut_in": 0.31},
    {"segment": "C03", "date": "2026-01-06", "crack_pct": 9, "rut_in": 0.12},
    {"segment": "C03", "date": "2026-01-06", "crack_pct": 10, "rut_in": 0.12},

    {"segment": "D21", "date": "2026-01-07", "crack_pct": 18, "rut_in": 0.22},
]

# Convert records into a DataFrame.
df = pd.DataFrame(records)

# Count exact duplicate rows.
exact_mask = df.duplicated(keep=False)
exact_count = int(exact_mask.sum())

# Define engineering keys for uniqueness.
key_cols = ["segment", "date"]

# Count repeated key combinations.
key_mask = df.duplicated(subset=key_cols, keep=False)
key_count = int(key_mask.sum())

# Remove only exact accidental copies.
clean_df = df.drop_duplicates().reset_index(drop=True)

# Flag remaining key conflicts for review.
review_mask = clean_df.duplicated(subset=key_cols, keep=False)
review_df = clean_df.loc[review_mask, key_cols + ["crack_pct"]]

# Compare averages before and after cleaning.
raw_mean = df["crack_pct"].mean()
clean_mean = clean_df["crack_pct"].mean()

# Print a compact cleaning summary.
print(f"Raw rows: {len(df)}")
print(f"Exact duplicate rows found: {exact_count}")
print(f"Rows sharing segment-date keys: {key_count}")
print(f"Rows after exact duplicate removal: {len(clean_df)}")
print(f"Mean cracking before cleaning: {raw_mean:.1f}%")
print(f"Mean cracking after cleaning: {clean_mean:.1f}%")
print("Potential conflicting duplicates for review:")
print(review_df.to_string(index=False))

```

### **1.3. Inconsistent Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1783553902" width="250">



>* Inconsistent values use mismatched formats or meanings
>* Standardize labels, units, and categories before analysis

>* Check categories, numbers, and dates for inconsistencies
>* Use engineering context to justify corrections

>* Judge inconsistencies before changing engineering data
>* Standardize, document, and improve dataset trust



```
#@title Python Code - Inconsistent Values

# Clean inconsistent civil engineering values.
# Standardize labels, units, and ratings.
# Keep outputs short for learners.

import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny inspection dataset.
records = [
    {"asset": "P1", "material": "concrete", "diameter": 24, "unit": "in", "rating": "Good"},
    {"asset": "P2", "material": "RCP", "diameter": 600, "unit": "mm", "rating": "G"},
    {"asset": "P3", "material": "conc", "diameter": 18, "unit": "in", "rating": "fair"},
    {"asset": "P4", "material": "PVC", "diameter": 45, "unit": "cm", "rating": "F"},
]

# Load records into a dataframe.
df = pd.DataFrame(records)

# Define controlled material labels.
material_map = {
    "concrete": "Reinforced concrete",
    "rcp": "Reinforced concrete",
    "conc": "Reinforced concrete",
    "pvc": "PVC",
}

# Define controlled condition labels.
rating_map = {
    "good": "Good",
    "g": "Good",
    "fair": "Fair",
    "f": "Fair",
}

# Standardize text before mapping.
df["material_key"] = df["material"].str.strip().str.lower()
df["rating_key"] = df["rating"].str.strip().str.lower()

# Convert all diameters to millimeters.
unit_factor = {"in": 25.4, "mm": 1.0, "cm": 10.0}
df["diameter_mm"] = df["diameter"] * df["unit"].map(unit_factor)

# Apply category standardization maps.
df["material_clean"] = df["material_key"].map(material_map)
df["rating_clean"] = df["rating_key"].map(rating_map)

# Check whether cleaning created missing values.
missing_clean = df[["material_clean", "rating_clean", "diameter_mm"]].isna().sum().sum()
assert missing_clean == 0, "Some values were not standardized."

# Print a compact before and after view.
print("Original material labels:", sorted(df["material"].unique()))
print("Clean material labels:", sorted(df["material_clean"].unique()))
print("Original units:", sorted(df["unit"].unique()))
print("Clean diameter range mm:", round(df["diameter_mm"].min(), 1), "to", round(df["diameter_mm"].max(), 1))
print("Clean ratings:", sorted(df["rating_clean"].unique()))

# Plot standardized diameters by asset.
plt.figure(figsize=(6, 3))
plt.bar(df["asset"], df["diameter_mm"], color="steelblue")
plt.ylabel("Diameter after cleaning, mm")
plt.title("Pipe diameters after unit harmonization")
plt.tight_layout()
plt.show()

```

## **2. Feature Preparation**

### **2.1. Feature Engineering Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1783553912" width="250">



>* Transform raw data into meaningful model features
>* Use engineering knowledge to reveal key patterns

>* Create features that reveal engineering relationships
>* Keep features simple, meaningful, and consistent

>* Create features with clear, relevant purpose
>* Avoid leakage and test feature stability



```
#@title Python Code - Feature Engineering Basics

# Feature engineering turns measurements into useful variables.
# Civil datasets often need simple derived features.
# Scaling helps compare variables with different units.

# Import small data and plotting libraries.
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny pavement inspection dataset.
data = {
    "section": ["A", "B", "C", "D", "E", "F"],
    "length_ft": [500, 800, 650, 900, 700, 550],

    "width_ft": [24, 24, 36, 30, 24, 28],
    "age_years": [4, 12, 8, 20, 15, 6],
    "daily_traffic": [9000, 18000, 14000, 26000, 21000, 11000],
    "heavy_truck_percent": [6, 12, 9, 18, 15, 8],

    "crack_length_ft": [120, 520, 310, 980, 760, 210],
}

# Convert the dictionary into a table.
df = pd.DataFrame(data)

# Create area from length and width.
df["area_sqft"] = df["length_ft"] * df["width_ft"]

# Estimate heavy trucks per day.
df["heavy_trucks_daily"] = (
    df["daily_traffic"] * df["heavy_truck_percent"] / 100
)

# Normalize cracking by pavement area.
df["crack_ft_per_1000_sqft"] = (
    df["crack_length_ft"] / df["area_sqft"] * 1000
)

# Combine age and heavy traffic exposure.
df["truck_age_exposure"] = (
    df["heavy_trucks_daily"] * df["age_years"]
)

# Scale selected features using min max scaling.
feature_cols = ["age_years", "heavy_trucks_daily", "crack_ft_per_1000_sqft"]

# Avoid division by zero during scaling.
for col in feature_cols:
    spread = df[col].max() - df[col].min()
    df[col + "_scaled"] = (df[col] - df[col].min()) / spread

# Select compact columns for display.
show_cols = ["section", "area_sqft", "heavy_trucks_daily"]
show_cols += ["crack_ft_per_1000_sqft", "age_years_scaled"]

# Print only a small teaching summary.
print("Engineered pavement features for modeling:")
print(df[show_cols].round(2).to_string(index=False))
print("Scaled values range from 0 to 1.")

# Plot one engineered relationship.
plt.figure(figsize=(7, 4))
plt.scatter(df["truck_age_exposure"], df["crack_ft_per_1000_sqft"], s=80)

# Label each pavement section.
for _, row in df.iterrows():
    plt.text(row["truck_age_exposure"], row["crack_ft_per_1000_sqft"], row["section"])

# Add clear engineering labels.
plt.xlabel("Heavy truck age exposure")
plt.ylabel("Crack length per 1000 square feet")
plt.title("Engineered features reveal pavement stress patterns")
plt.grid(True, alpha=0.3)

# Display the single plot.
plt.show()

```

### **2.2. Feature Scaling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1783553914" width="250">



>* Put different measurements on comparable scales
>* Prevent large numbers from dominating models

>* Standardization compares values to typical spread
>* Normalization rescales values to fixed ranges

>* Fit scaling on training data only
>* Check outliers with engineering judgment



```
#@title Python Code - Feature Scaling

# Feature scaling prepares engineering variables.
# Small examples make transformations clear.
# Plots compare original and scaled ranges.

# Import common data and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny pavement condition dataset.
data = {
    "asphalt_thickness_mm": [80, 95, 110, 125, 140, 155],
    "traffic_vehicles_day": [4200, 7600, 11800, 15300, 20100, 24800],
    "rainfall_mm_year": [520, 610, 690, 760, 840, 930],
    "cost_thousand_usd": [180, 225, 260, 310, 370, 430],
}

# Store measurements in a dataframe.
df = pd.DataFrame(data)

# Validate the dataset before scaling.
assert df.shape == (6, 4)
assert df.notna().all().all()

# Select numeric engineering features.
features = df.columns.tolist()
values = df[features].astype(float)

# Compute standardization using training-style rules.
means = values.mean()
stds = values.std(ddof=0)
standardized = (values - means) / stds

# Compute min-max normalization safely.
mins = values.min()
ranges = values.max() - values.min()
normalized = (values - mins) / ranges

# Add one useful engineered feature.
df["traffic_per_mm_thickness"] = (
    df["traffic_vehicles_day"] / df["asphalt_thickness_mm"]
)

# Print compact summaries for learners.
print("Original ranges:")
print((values.max() - values.min()).round(1).to_dict())
print("Standardized means:")
print(standardized.mean().round(2).to_dict())
print("Normalized min and max:")
print(normalized.agg(["min", "max"]).round(2).to_dict())
print("Engineered feature first value:")
print(round(df["traffic_per_mm_thickness"].iloc[0], 2))

# Prepare one plot comparing scaling methods.
plot_data = pd.DataFrame({
    "original_traffic": values["traffic_vehicles_day"],
    "standardized_traffic": standardized["traffic_vehicles_day"],
    "normalized_traffic": normalized["traffic_vehicles_day"],
})

# Draw compact line plots for comparison.
ax = plot_data.plot(marker="o", figsize=(8, 4))
ax.set_title("Traffic Feature Before and After Scaling")
ax.set_xlabel("Road segment index")
ax.set_ylabel("Feature value")

# Improve readability for the single figure.
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **2.3. Feature Scaling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1783553910" width="250">



>* Put different measurements on comparable scales
>* Prevent large values from unfairly dominating

>* Standardize around typical values and spread
>* Normalize to bounded, meaningful ranges

>* Fit scaling only on training data
>* Scale thoughtfully while preserving engineering meaning



```
#@title Python Code - Feature Scaling

# Feature scaling supports fair engineering comparisons.
# Small pavement data keeps calculations transparent.
# Manual formulas avoid machine learning libraries.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a tiny civil engineering dataset.
data = {
    "thickness_mm": [90, 110, 105, 130, 95, 125],
    "traffic_vpd": [8000, 15000, 12000, 22000, 9000, 18000],
    "roughness_m_per_km": [1.8, 2.4, 2.1, 3.2, 1.9, 2.8],
}

# Store measurements in a dataframe.
df = pd.DataFrame(data)

# Select numeric features for scaling.
features = ["thickness_mm", "traffic_vpd", "roughness_m_per_km"]
raw = df[features].copy()

# Compute standardization parameters from available training data.
means = raw.mean()
stds = raw.std(ddof=0)
standardized = (raw - means) / stds

# Compute normalization parameters from available training data.
mins = raw.min()
ranges = raw.max() - raw.min()
normalized = (raw - mins) / ranges

# Add one useful engineered feature before scaling.
df["traffic_per_mm"] = df["traffic_vpd"] / df["thickness_mm"]

# Print compact summaries for learners.
print("Raw feature ranges:")
print((raw.max() - raw.min()).round(2).to_dict())
print("Standardized means:", standardized.mean().round(2).to_dict())
print("Normalized min/max:", normalized.min().min(), normalized.max().max())
print("Engineered feature example:", round(df.loc[0, "traffic_per_mm"], 2))

# Compare raw and scaled traffic values visually.
plot_df = pd.DataFrame({
    "raw_traffic": raw["traffic_vpd"],
    "standardized_traffic": standardized["traffic_vpd"],
    "normalized_traffic": normalized["traffic_vpd"],
})

# Draw exactly one compact plot.
ax = plot_df.plot(kind="bar", figsize=(8, 4))
ax.set_title("Traffic volume before and after scaling")
ax.set_xlabel("Pavement section")
ax.set_ylabel("Displayed value")

# Improve readability without extra printed output.
plt.tight_layout()
plt.show()

```

### **3.1. Distribution Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1783553904" width="250">

>* Histograms reveal distribution shape and unusual patterns

>* Compare groups using medians, spread, and outliers

```
#@title Python Code - Distribution Plots

# Distribution plots summarize engineering measurement spread.
# Histograms and boxplots reveal unusual values.
# This example uses small concrete data.

# Import numerical, table, and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make the random example reproducible.
rng = np.random.default_rng(seed=42)

# Create realistic concrete strength measurements.
strength_mpa = rng.normal(loc=34, scale=4, size=36)
strength_mpa = np.append(strength_mpa, [22, 47])

# Store values in a small dataframe.
data = pd.DataFrame({"strength_mpa": strength_mpa})

# Add a scaled feature for later modeling.
data["strength_scaled"] = (
    data["strength_mpa"] - data["strength_mpa"].mean()
) / data["strength_mpa"].std()

# Check the dataset before plotting.
assert data.shape[0] >= 10 and data.shape[1] == 2

# Print a compact engineering summary.
print("Concrete strength distribution example")
print(f"Samples: {len(data)} cylinders")
print(f"Mean: {data['strength_mpa'].mean():.1f} MPa")
print(f"Median: {data['strength_mpa'].median():.1f} MPa")
print(f"Minimum: {data['strength_mpa'].min():.1f} MPa")
print(f"Maximum: {data['strength_mpa'].max():.1f} MPa")

# Create one figure with two distribution views.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot a histogram for overall shape.
axes[0].hist(
    data["strength_mpa"], bins=8, color="steelblue", edgecolor="black"
)

# Label the histogram clearly.
axes[0].set_title("Histogram of Concrete Strength")
axes[0].set_xlabel("Compressive strength (MPa)")
axes[0].set_ylabel("Number of cylinders")

# Plot a boxplot for spread and outliers.
axes[1].boxplot(
    data["strength_mpa"], vert=True, patch_artist=True
)

# Label the boxplot clearly.
axes[1].set_title("Box Plot of Concrete Strength")
axes[1].set_ylabel("Compressive strength (MPa)")
axes[1].set_xticks([1], ["Batch A"])

# Improve spacing before displaying the figure.
plt.tight_layout()
plt.show()

```

### **3.2. Time Trend Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1783553906" width="250">

>* Use line plots to reveal trends

>* Use readable axes, units, titles, and labels

```
#@title Python Code - Time Trend Plots

# Time trend plots reveal engineering changes.
# Small datasets make cleaning steps visible.
# Matplotlib lines show direction over time.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a small bridge monitoring dataset.
dates = pd.date_range("2026-01-01", periods=14, freq="D")
strain = [102, 104, 103, np.nan, 108, 108, 111]
strain = strain + [115, 160, 118, 119, 121, 122, 123]

# Build the dataframe with one duplicate row.
data = pd.DataFrame({"date": dates, "strain_microstrain": strain})
duplicate = data.iloc[[5]].copy()
data = pd.concat([data, duplicate], ignore_index=True)

# Clean duplicate dates and missing readings.
data = data.drop_duplicates(subset="date", keep="first")
data["strain_microstrain"] = data["strain_microstrain"].interpolate()
data = data.sort_values("date").reset_index(drop=True)

# Create simple time features for modeling later.
data["day_number"] = np.arange(1, len(data) + 1)
data["strain_scaled"] = data["strain_microstrain"] / 100.0

# Print a compact summary for learners.
print("Rows after cleaning:", len(data))
print("Missing strain values:", int(data["strain_microstrain"].isna().sum()))
print("Largest strain:", float(data["strain_microstrain"].max()))

# Prepare one clear time trend plot.
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(data["date"], data["strain_microstrain"], marker="o")
ax.axhline(140, color="red", linestyle="--", label="Review threshold")

# Label the engineering context clearly.
ax.set_title("Bridge Strain Sensor Time Trend")
ax.set_xlabel("Date")
ax.set_ylabel("Strain, microstrain")
ax.legend()

# Improve date readability and spacing.
fig.autofmt_xdate()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **3.3. Grid Pattern Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1783553908" width="250">



>* Show measurements across physical site layouts
>* Reveal spatial patterns and problem zones

>* Color shading reveals spatial value patterns
>* Use clear scales, legends, and labels

>* Avoid implying false precision or complete coverage
>* Use grids to guide engineering decisions



```
#@title Python Code - Grid Pattern Plots

# Grid plots show spatial engineering measurements.
# Colors reveal patterns across physical locations.
# This example maps bridge deck cover.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create a small measured grid.
cover_cm = np.array([
    [6.2, 6.0, 5.8, 5.5, 5.7],
    [6.4, 5.9, 4.8, 4.6, 5.4],
    [6.1, 5.6, 4.2, 4.4, 5.2],

    [6.3, 5.8, 4.9, 5.0, 5.6],
    [6.5, 6.1, 5.7, 5.9, 6.0],
])

# Validate the grid before plotting.
rows, cols = cover_cm.shape
assert rows == 5 and cols == 5

# Define engineering threshold value.
minimum_cover_cm = 5.0
low_cells = np.sum(cover_cm < minimum_cover_cm)

# Print a short interpretation summary.
print("Bridge deck cover grid: 5 by 5 cells")
print(f"Minimum measured cover: {cover_cm.min():.1f} cm")
print(f"Cells below {minimum_cover_cm:.1f} cm: {low_cells}")

# Create one grid pattern plot.
fig, ax = plt.subplots(figsize=(6, 5))
image = ax.imshow(cover_cm, cmap="YlOrRd_r", vmin=4, vmax=7)

# Add a color legend.
colorbar = fig.colorbar(image, ax=ax)
colorbar.set_label("Concrete cover depth (cm)")

# Label physical grid directions.
ax.set_title("Bridge Deck Concrete Cover Grid")
ax.set_xlabel("Transverse position across deck")
ax.set_ylabel("Longitudinal station along deck")

# Show cell labels for beginners.
for row in range(rows):
    for col in range(cols):
        value = cover_cm[row, col]
        ax.text(col, row, f"{value:.1f}", ha="center", va="center")

# Mark the low-cover threshold visually.
ax.contour(cover_cm < minimum_cover_cm, levels=[0.5], colors="blue")
plt.tight_layout()
plt.show()

```
