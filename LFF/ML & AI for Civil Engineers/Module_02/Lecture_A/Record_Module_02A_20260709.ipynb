## **1. NumPy Arrays**

### **1.1. Array Creation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1783611729" width="250">



>* Organize engineering measurements into NumPy arrays
>* Enable fast, consistent numerical analysis

>* Create arrays from collected measurement sequences
>* Organize structured data into rows and columns

>* Generate arrays for models and simulations
>* Organize data for efficient engineering computation



```
#@title Python Code - Array Creation

# Create arrays for engineering measurements.
# Use small examples for clarity.
# Inspect shapes before later calculations.

# NumPy is already available in Colab.
import numpy as np

# Create one dimensional settlement readings.
settlement_mm = np.array([12.4, 13.1, 11.8, 14.0])

# Create two dimensional girder deflection data.
deflection_in = np.array([[0.12, 0.18, 0.15], [0.10, 0.16, 0.14]])

# Generate evenly spaced station distances.
stations_m = np.linspace(0, 100, 6)

# Generate an initial groundwater head grid.
head_grid_m = np.full((3, 4), 25.0)

# Validate expected array shapes.
assert settlement_mm.shape == (4,)
assert deflection_in.shape == (2, 3)
assert head_grid_m.shape == (3, 4)

# Print compact summaries only.
print("Settlement array:", settlement_mm)
print("Settlement shape:", settlement_mm.shape)
print("Deflection grid shape:", deflection_in.shape)
print("Station distances:", stations_m)
print("Head grid shape:", head_grid_m.shape)
print("Head grid mean:", head_grid_m.mean())

```

### **1.2. Array Attributes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1783611727" width="250">



>* Attributes reveal array structure and data type

>* Array shape shows how values are arranged

```
#@title Python Code - Array Attributes

# Inspect array attributes for engineering measurements.
# Shapes reveal physical data organization.
# Data types affect precision and memory.

# Import NumPy for numerical array work.
import numpy as np

# Create hourly rainfall depths in inches.
rainfall_inches = np.array([0.0, 0.2, 0.5, 0.1])

# Create bridge deflection grid in millimeters.
deflection_mm = np.array([[1.2, 1.5, 1.4], [1.8, 2.0, 1.9]])

# Validate the expected grid layout.
expected_shape = (2, 3)

# Check shape before interpreting positions.
shape_is_expected = deflection_mm.shape == expected_shape

# Print compact attributes for both arrays.
print("Rainfall ndim, shape, size:", rainfall_inches.ndim, rainfall_inches.shape, rainfall_inches.size)

# Print compact attributes for grid data.
print("Deflection ndim, shape, size:", deflection_mm.ndim, deflection_mm.shape, deflection_mm.size)

# Print data types and memory use.
print("Rainfall dtype and bytes:", rainfall_inches.dtype, rainfall_inches.nbytes)

# Print grid type and layout check.
print("Deflection dtype and shape ok:", deflection_mm.dtype, shape_is_expected)

# Summarize engineering interpretation safely.
print("Grid means 2 girders by 3 measurement points.")

```

### **1.3. Grid Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1783611724" width="250">



>* Grid arrays organize measurements by location
>* Position gives engineering values important meaning

>* Apply transformations across entire grids consistently
>* Select grid regions for focused analysis

>* Grid arrays support modeling and visualization workflows
>* Context gives grid values engineering meaning



```
#@title Python Code - Grid Data

# NumPy grids store engineering measurements.
# Rows and columns preserve locations.
# Small arrays make patterns visible.

# numpy is preinstalled in Google Colab.
import numpy as np
import matplotlib.pyplot as plt

# Create a small settlement grid.
settlement_mm = np.array([
    [4.0, 5.5, 6.0, 5.0],
    [5.0, 8.5, 12.0, 7.5],
    [4.5, 7.0, 9.5, 6.5],
])

# Validate the expected grid shape.
expected_shape = (3, 4)
assert settlement_mm.shape == expected_shape

# Convert millimeters to inches.
settlement_in = settlement_mm / 25.4
max_mm = settlement_mm.max()

# Select one row across foundation.
centerline_mm = settlement_mm[1, :]
centerline_mean = centerline_mm.mean()

# Find cells above design threshold.
threshold_mm = 8.0
high_mask = settlement_mm > threshold_mm
high_count = np.count_nonzero(high_mask)

# Print compact engineering summaries.
print("Grid shape:", settlement_mm.shape)
print("Maximum settlement mm:", round(max_mm, 2))
print("Centerline mean mm:", round(centerline_mean, 2))
print("Cells above threshold:", int(high_count))
print("Maximum settlement inches:", round(settlement_in.max(), 3))

# Plot the grid as a heatmap.
plt.figure(figsize=(5, 3))
plt.imshow(settlement_mm, cmap="YlOrRd")
plt.colorbar(label="Settlement (mm)")
plt.title("Foundation Settlement Grid")
plt.xlabel("Column position")
plt.ylabel("Row position")
plt.tight_layout()
plt.show()

```

## **2. pandas DataFrames**

### **2.1. Loading DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1783611717" width="250">



>* Load engineering records into structured DataFrames
>* Use named columns for systematic analysis

>* Load varied engineering files into pandas
>* Check headers, dates, missing values, numbers

>* Inspect loaded data for quality issues
>* Verify context before analysis and decisions



```
#@title Python Code - Loading DataFrames

# Load civil engineering records into pandas.
# Inspect columns before deeper analysis.
# Keep outputs short for beginners.

# pandas and numpy are preinstalled in Colab.
import pandas as pd
from io import StringIO

# Store tiny traffic data as CSV text.
csv_text = """date,intersection,direction,hour,vehicles,delay_seconds
2026-01-05,Main & 1st,NB,7,420,38
2026-01-05,Main & 1st,SB,7,390,35
2026-01-05,Main & 1st,NB,8,510,52
2026-01-05,Oak & 3rd,EB,7,275,24
2026-01-05,Oak & 3rd,WB,8,310,29
"""

# Read CSV text into a DataFrame.
traffic_df = pd.read_csv(
    StringIO(csv_text),
    parse_dates=["date"]
)

# Confirm the loaded table size.
row_count, column_count = traffic_df.shape
print(f"Loaded rows: {row_count}, columns: {column_count}")

# Show column names for quick checking.
column_names = list(traffic_df.columns)
print("Columns:", column_names)

# Inspect only the first few records.
print("First two rows:")
print(traffic_df.head(2).to_string(index=False))

# Filter records for heavier traffic.
heavy_df = traffic_df[traffic_df["vehicles"] >= 400]
print("Heavy traffic rows:", len(heavy_df))

# Summarize vehicle counts by intersection.
summary = traffic_df.groupby("intersection")["vehicles"].mean()
print("Average vehicles by intersection:")
print(summary.round(1).to_string())

```

### **2.2. Filtering DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1783611719" width="250">



>* Select relevant rows or columns for analysis
>* Create focused subsets for engineering decisions

>* Filter rows using column-based conditions
>* Combine criteria to focus engineering analysis

>* Check data quality before filtering
>* Match filters to engineering context



```
#@title Python Code - Filtering DataFrames

# Filtering selects rows matching engineering questions.
# Small DataFrames make conditions easy to inspect.
# Civil datasets often need focused subsets.

# Import pandas for tabular engineering data.
import pandas as pd

# Create a small pavement inspection table.
data = {
    "section": ["A1", "A2", "B1", "B2", "C1", "C2"],
    "material": ["asphalt", "concrete", "asphalt", "asphalt", "concrete", "asphalt"],

    "year": [2023, 2024, 2024, 2022, 2024, 2024],
    "iri_in_per_mile": [95, 130, 160, 110, 145, 88],
    "rut_depth_in": [0.18, 0.05, 0.32, 0.22, 0.04, 0.12],
}

# Convert the dictionary into a DataFrame.
df = pd.DataFrame(data)

# Build clear filtering conditions separately.
is_asphalt = df["material"].eq("asphalt")
is_recent = df["year"].eq(2024)
needs_review = df["iri_in_per_mile"].gt(100)

# Combine conditions using logical and.
filtered = df.loc[is_asphalt & is_recent & needs_review]

# Select only columns needed for review.
review_columns = ["section", "iri_in_per_mile", "rut_depth_in"]
filtered_view = filtered.loc[:, review_columns]

# Summarize the filtered engineering subset.
count_sections = len(filtered_view)
average_iri = filtered_view["iri_in_per_mile"].mean()

# Print compact results for learners.
print("Original rows:", len(df))
print("Filtered rows:", count_sections)
print("Average filtered IRI:", round(average_iri, 1))
print(filtered_view.to_string(index=False))

```

### **2.3. Summarizing DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1783611721" width="250">



>* Condense large engineering tables into useful summaries
>* Check data patterns before deeper analysis

>* Summarize columns by data type
>* Find patterns, extremes, and time gaps

>* Group data to compare engineering conditions
>* Use summaries for decisions and quality checks



```
#@title Python Code - Summarizing DataFrames

# Summaries turn tables into engineering insight.
# Small DataFrames are ideal for learning.
# Grouped summaries support practical comparisons.

# Import pandas for tabular engineering data.
import pandas as pd

# Create a small concrete testing dataset.
records = [
    {"mix": "A", "age_days": 7, "strength_mpa": 28.4, "slump_cm": 9.0},
    {"mix": "A", "age_days": 28, "strength_mpa": 41.2, "slump_cm": 8.5},

    {"mix": "B", "age_days": 7, "strength_mpa": 31.1, "slump_cm": 10.0},
    {"mix": "B", "age_days": 28, "strength_mpa": 45.8, "slump_cm": 9.5},
    {"mix": "B", "age_days": 28, "strength_mpa": None, "slump_cm": 9.2},
]

# Convert records into a pandas DataFrame.
df = pd.DataFrame(records)

# Inspect size and column names briefly.
print("Rows and columns:", df.shape)
print("Column names:", list(df.columns))

# Summarize numeric columns with common statistics.
summary = df.describe().round(2)
print("Numeric summary:")
print(summary.loc[["count", "mean", "min", "max"]])

# Count missing values in each column.
missing_counts = df.isna().sum()
print("Missing values:", missing_counts.to_dict())

# Summarize strength by concrete mix.
grouped = df.groupby("mix")["strength_mpa"].agg(["count", "mean", "max"])
print("Strength by mix:")
print(grouped.round(2))

# Filter records meeting a design threshold.
passing = df[df["strength_mpa"] >= 40]
print("Tests at least 40 MPa:", len(passing))

```

## **3. Data Conversion**

### **3.1. List Conversions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1783611731" width="250">



>* Lists support flexible, gradual data collection
>* Convert lists for efficient numerical analysis

>* Arrays enable fast numerical measurement analysis
>* Consistent data types support shaped grid data

>* Choose arrays for uniform numerical calculations
>* Choose DataFrames for labeled mixed records



```
#@title Python Code - List Conversions

# Convert lists into analysis structures.
# Civil measurements often start simply.
# Arrays and tables support analysis.

# numpy and pandas are preinstalled in Colab.
import numpy as np
import pandas as pd

# Start with a simple engineering list.
strengths_psi = [4100, 4350, 3980, 4520]
strength_array = np.array(strengths_psi, dtype=float)

# Convert psi strengths into megapascals.
strengths_mpa = strength_array * 0.00689476
average_mpa = strengths_mpa.mean()

# Build a labeled table from lists.
specimen_ids = ["C1", "C2", "C3", "C4"]
concrete_df = pd.DataFrame({"id": specimen_ids, "strength_psi": strengths_psi})

# Add the converted NumPy values.
concrete_df["strength_mpa"] = strengths_mpa.round(2)
passing_df = concrete_df[concrete_df["strength_psi"] >= 4000]

# Convert a nested list into grid data.
elevation_grid_ft = [[101.2, 101.5], [100.9, 101.1]]
elevation_array = np.array(elevation_grid_ft, dtype=float)

# Validate the expected grid shape.
expected_shape = (2, 2)
assert elevation_array.shape == expected_shape

# Print compact results for learners.
print("Original list:", strengths_psi)
print("NumPy average MPa:", round(average_mpa, 2))
print("DataFrame preview:")
print(concrete_df.head(4).to_string(index=False))
print("Passing specimen count:", len(passing_df))
print("Elevation grid shape:", elevation_array.shape)

```

### **3.2. DataFrame Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1783611733" width="250">



>* DataFrames organize labeled engineering table data
>* Structured data enables filtering and summaries

>* Combine varied engineering data into labeled tables
>* Improve interpretation, accuracy, and team communication

>* Convert DataFrames for faster numerical computation
>* Choose structures that fit each task



```
#@title Python Code - DataFrame Conversion

# Convert engineering containers into labeled tables.
# Use arrays when numerical computation matters.
# Use DataFrames when labels matter.

# Import numerical and tabular tools.
import numpy as np
import pandas as pd

# Store simple field measurements in lists.
segment_ids = ["A1", "A2", "A3"]
length_ft = [120.0, 150.0, 90.0]
condition = [82, 74, 91]

# Convert lists into a labeled DataFrame.
road_df = pd.DataFrame({
    "segment_id": segment_ids,
    "length_ft": length_ft,
    "condition_score": condition,
})

# Inspect only a small table preview.
print("DataFrame from Python lists:")
print(road_df.head(3).to_string(index=False))

# Select numeric columns for array computation.
numeric_cols = ["length_ft", "condition_score"]
road_array = road_df[numeric_cols].to_numpy()
assert road_array.shape == (3, 2)

# Convert feet to meters using NumPy.
length_m = road_array[:, 0] * 0.3048
mean_length_m = length_m.mean()

# Convert computed array results back into DataFrame.
road_df["length_m"] = np.round(length_m, 2)
print("\nMean segment length in meters:")
print(round(mean_length_m, 2))

# Convert selected DataFrame columns into records.
records = road_df[["segment_id", "length_m"]].to_dict("records")
print("\nReport-friendly records:")
print(records)

# Summarize why each structure was useful.
summary = "Lists organize input, arrays compute, DataFrames label results."
print("\nKey idea:")
print(summary)

```

### **3.3. Container Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1783611735" width="250">



>* Convert containers for different data tasks
>* Use arrays for math, DataFrames for labels

>* Use arrays for same-type measurements
>* Use DataFrames for structured field data

>* Keep labels and context during conversions
>* Simplify data only when meaning stays clear



```
#@title Python Code - Container Conversion

# Convert containers for engineering data.
# Use arrays for numerical calculations.
# Use DataFrames for labeled tables.

# Import numerical and tabular libraries.
import numpy as np
import pandas as pd

# Start with a simple Python list.
strengths_psi = [3920, 4050, 3985, 4120]
strength_array = np.array(strengths_psi, dtype=float)

# Convert psi values into megapascals.
strength_mpa = strength_array * 0.00689476
mean_mpa = strength_mpa.mean()

# Build a dictionary for tabular data.
test_data = {
    "cylinder": ["A", "B", "C", "D"],
    "strength_psi": strengths_psi,
    "strength_mpa": strength_mpa.round(2),
}

# Convert the dictionary into DataFrame.
df = pd.DataFrame(test_data)
strong_df = df[df["strength_psi"] >= 4000]

# Convert one DataFrame column back.
selected_list = strong_df["cylinder"].tolist()
selected_array = strong_df["strength_mpa"].to_numpy()

# Show compact conversion results.
print("Original container:", type(strengths_psi).__name__)
print("Array shape:", strength_array.shape)
print("Mean strength MPa:", round(mean_mpa, 2))
print("DataFrame columns:", list(df.columns))
print("Strong cylinders:", selected_list)
print("Selected MPa array:", selected_array.round(2))

```
