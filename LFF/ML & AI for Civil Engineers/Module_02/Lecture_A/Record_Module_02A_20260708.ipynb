## **1. NumPy Arrays**

### **1.1. Vector Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1783552881" width="250">



>* Engineering measurements often form one-dimensional vectors
>* NumPy arrays enable consistent batch calculations

>* Apply operations to entire measurement vectors
>* Ensure consistent processing for engineering analysis

>* Ordered vectors preserve engineering measurement meaning
>* Consistent arrays support advanced data analysis



```
#@title Python Code - Vector Measurements

# NumPy stores engineering measurements as vectors.
# Vector operations apply consistently to every value.
# This example uses concrete strength measurements.

# Import NumPy for numerical array operations.
import numpy as np

# Store compressive strengths measured in megapascals.
strength_mpa = np.array([31.2, 34.8, 29.5, 36.1, 33.7])

# Validate the vector has one dimension.
assert strength_mpa.ndim == 1 and strength_mpa.size == 5

# Convert every measurement from megapascals to psi.
strength_psi = strength_mpa * 145.038

# Compare every value against a design threshold.
design_mpa = 32.0
passes_design = strength_mpa >= design_mpa

# Compute simple summary values for the vector.
average_mpa = np.mean(strength_mpa)
minimum_mpa = np.min(strength_mpa)
maximum_mpa = np.max(strength_mpa)

# Print compact results for classroom inspection.
print("Strength vector shape:", strength_mpa.shape)
print("Strengths in MPa:", strength_mpa)
print("Strengths in psi:", np.round(strength_psi, 1))
print("Passes 32 MPa design:", passes_design)
print("Average, minimum, maximum MPa:", round(average_mpa, 2), minimum_mpa, maximum_mpa)

```

### **1.2. Array Creation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1783552879" width="250">



>* Organize measurements into structured NumPy arrays
>* Apply calculations efficiently across engineering data

>* Create arrays from known engineering data
>* Choose shapes that preserve data meaning

>* Create placeholder and evenly spaced arrays
>* Choose shapes for reliable engineering calculations



```
#@title Python Code - Array Creation

# Create arrays for engineering measurements.
# Organize sequences and grid data.
# Inspect shapes before later calculations.

# NumPy is preinstalled in Google Colab.
import numpy as np

# Create a one dimensional measurement array.
strength_mpa = np.array([31.2, 34.8, 33.5, 36.1])

# Create evenly spaced roadway stations.
stations_m = np.linspace(0, 300, 4)

# Create a two dimensional sensor table.
deflection_in = np.array([[0.02, 0.03, 0.04], [0.01, 0.02, 0.03]])

# Create placeholder predictions using zeros.
predicted_mm = np.zeros((2, 3))

# Create a constant load grid.
uniform_load_kpa = np.full((2, 3), 12.5)

# Validate expected array shapes.
assert strength_mpa.shape == (4,)
assert deflection_in.shape == predicted_mm.shape

# Print compact summaries only.
print("Strength array:", strength_mpa)
print("Stations in meters:", stations_m)
print("Deflection grid shape:", deflection_in.shape)
print("Zero prediction grid:\n", predicted_mm)
print("Uniform load grid:\n", uniform_load_kpa)
print("Maximum strength MPa:", strength_mpa.max())

```

### **1.3. Grid Data Arrays**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1783552883" width="250">



>* Grid arrays organize measurements by position
>* Spatial layout supports meaningful engineering calculations

>* Rows and columns organize engineering measurements
>* NumPy quickly calculates across entire grids

>* Grids support visualization, simulation, and analysis.
>* Track dimensions for correct engineering interpretation.



```
#@title Python Code - Grid Data Arrays

# Create grid arrays for engineering measurements.
# Rows and columns preserve spatial meaning.
# NumPy summarizes grids without manual loops.

# Import NumPy for numerical grid arrays.
import numpy as np

# Store site elevations in meters.
elevation_m = np.array([
    [101.2, 101.5, 101.9],
    [100.8, 101.1, 101.6],

    [100.3, 100.7, 101.0]
])

# Validate the grid has expected dimensions.
expected_shape = (3, 3)
assert elevation_m.shape == expected_shape

# Compute useful engineering summaries.
mean_elevation = elevation_m.mean()
max_elevation = elevation_m.max()

# Compare rows as north-south survey lines.
row_means = elevation_m.mean(axis=1)

# Compare columns as west-east survey lines.
column_means = elevation_m.mean(axis=0)

# Convert elevations from meters to feet.
elevation_ft = elevation_m * 3.28084

# Print compact results for learners.
print("Elevation grid shape:", elevation_m.shape)
print("Mean elevation, meters:", round(mean_elevation, 2))
print("Maximum elevation, meters:", round(max_elevation, 2))
print("Row means, meters:", np.round(row_means, 2))
print("Column means, meters:", np.round(column_means, 2))
print("Center point, feet:", round(elevation_ft[1, 1], 2))

```

## **2. pandas DataFrames**

### **2.1. Loading DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1783552885" width="250">



>* DataFrames organize engineering data into rows and columns
>* Loaded data can be inspected and prepared

>* Check formats, units, headers, and missing values
>* Verify pandas interpreted the data correctly

>* Load records into decision-ready DataFrames
>* Check sources, columns, units, and record counts



```
#@title Python Code - Loading DataFrames

# Load small civil engineering tables.
# Inspect columns before deeper analysis.
# Keep output short and readable.

# pandas is preinstalled in Google Colab.
import pandas as pd
from io import StringIO

# Store tiny bridge inspection data.
csv_text = """bridge_id,county,span_ft,condition_score,inspection_date
B-101,North,120,7,2024-01-15
B-102,East,85,5,2024-02-03
B-103,North,210,,2024-02-20
B-104,West,60,8,2024-03-11
"""

# Load text as a DataFrame.
bridge_df = pd.read_csv(StringIO(csv_text))

# Convert dates after loading.
bridge_df["inspection_date"] = pd.to_datetime(bridge_df["inspection_date"])

# Check expected table size.
expected_columns = {"bridge_id", "county", "span_ft", "condition_score"}
assert expected_columns.issubset(bridge_df.columns)

# Print compact loading checks.
print("Loaded shape:", bridge_df.shape)
print("Column names:", list(bridge_df.columns))
print("Data types:", bridge_df.dtypes.astype(str).to_dict())

# Preview only the first records.
print("First two rows:")
print(bridge_df.head(2).to_string(index=False))

# Summarize missing values briefly.
print("Missing values:", bridge_df.isna().sum().to_dict())

# Filter records needing attention.
needs_review = bridge_df[bridge_df["condition_score"].fillna(0) < 6]
print("Needs review IDs:", needs_review["bridge_id"].tolist())

```

### **2.2. Filtering DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1783552888" width="250">



>* Select relevant rows or columns
>* Compare focused subsets with engineering criteria

>* Keep rows matching column conditions
>* Combine filters for focused engineering analysis

>* Check data quality before filtering rows
>* Use engineering judgment to interpret filtered results



```
#@title Python Code - Filtering DataFrames

# Filtering keeps engineering tables focused.
# Conditions select rows matching questions.
# Small examples make pandas clearer.

import pandas as pd

# Create a small bridge inspection table.
data = {
    "bridge_id": ["B101", "B102", "B103", "B104", "B105", "B106"],
    "county": ["Ada", "Ada", "Canyon", "Ada", "Canyon", "Boise"],
    "deck_rating": [7, 4, 5, 3, 8, 4],
    "span_ft": [120, 85, 210, 60, 150, 95],
}

# Convert the dictionary into DataFrame.
bridges = pd.DataFrame(data)

# Build one condition for poor decks.
poor_deck = bridges["deck_rating"] <= 4

# Build another condition for Ada county.
ada_county = bridges["county"] == "Ada"

# Combine conditions using pandas operators.
priority = bridges[poor_deck & ada_county]

# Select only columns needed for reporting.
report = priority[["bridge_id", "deck_rating", "span_ft"]]

# Summarize the filtered subset safely.
count_priority = len(report)
average_span = report["span_ft"].mean()

# Print compact results for learners.
print("Original rows:", len(bridges))
print("Priority Ada bridges:", count_priority)
print("Average priority span, ft:", round(average_span, 1))
print(report.to_string(index=False))

```

### **2.3. Summarizing DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1783552886" width="250">



>* Condense large engineering tables into useful summaries
>* Check data structure, completeness, and plausibility

>* Summarize numerical and categorical engineering data
>* Spot patterns, gaps, and data quality issues

>* Group data to compare engineering subsets
>* Summaries support evidence-based engineering decisions



```
#@title Python Code - Summarizing DataFrames

# Summaries turn engineering tables into useful evidence.
# This example uses small bridge inspection records.
# We inspect, filter, group, and summarize.

# Import pandas for DataFrame analysis.
import pandas as pd

# Create a tiny civil engineering dataset.
bridge_data = {
    "bridge_id": ["B01", "B02", "B03", "B04", "B05", "B06"],
    "district": ["North", "North", "South", "South", "East", "East"],

    "condition": ["Good", "Fair", "Poor", "Fair", "Good", "Poor"],
    "span_ft": [120, 85, 210, 160, 95, 180],
    "age_years": [12, 35, 58, 41, 18, 64],

    "daily_traffic": [18000, 9500, 22000, 14000, 7600, 19500],
    "deck_rating": [8, 6, 4, 5, 7, 3],
}

# Convert the dictionary into a DataFrame.
bridges = pd.DataFrame(bridge_data)

# Validate the table before summarizing.
assert bridges.shape == (6, 7)
assert bridges["deck_rating"].notna().all()

# Print compact inspection information.
print("Rows and columns:", bridges.shape)
print("Column names:", list(bridges.columns))

# Summarize selected numerical engineering measurements.
numeric_summary = bridges[["span_ft", "age_years", "daily_traffic", "deck_rating"]].describe()
print("Mean values:", numeric_summary.loc["mean"].round(1).to_dict())

# Count categorical condition ratings.
condition_counts = bridges["condition"].value_counts().to_dict()
print("Condition counts:", condition_counts)

# Filter bridges needing closer engineering review.
review_bridges = bridges[bridges["deck_rating"] <= 5]
print("Review bridge IDs:", review_bridges["bridge_id"].tolist())

# Group by district and summarize traffic.
district_summary = bridges.groupby("district")["daily_traffic"].mean().round(0)
print("Average traffic by district:", district_summary.to_dict())

```

## **3. Data Conversion**

### **3.1. Lists to Arrays**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1783552890" width="250">



>* Lists capture raw engineering measurements easily
>* Arrays enable efficient numerical analysis

>* Arrays require consistent measurement types and units
>* NumPy enables clean calculations on repeated data

>* Convert grid-like lists for numerical analysis
>* Use tables for mixed descriptive data



```
#@title Python Code - Lists to Arrays

# Lists often start engineering measurement workflows.
# Arrays support consistent numerical engineering calculations.
# This example converts settlement lists safely.

# Import NumPy for numerical array operations.
import numpy as np

# Store foundation settlement readings in millimeters.
settlement_list = [2.1, 2.4, 2.9, 3.2]

# Convert the Python list into a NumPy array.
settlement_array = np.array(settlement_list, dtype=float)

# Validate the array before engineering calculations.
assert settlement_array.ndim == 1 and settlement_array.size == 4

# Compute changes between consecutive monitoring visits.
settlement_changes = np.diff(settlement_array)

# Convert millimeters to inches for reporting.
settlement_inches = settlement_array / 25.4

# Create grid-like elevation data from nested lists.
elevation_grid_list = [[101.2, 101.5], [100.9, 101.1]]

# Convert nested lists into a two-dimensional array.
elevation_grid_array = np.array(elevation_grid_list, dtype=float)

# Validate the grid shape before selecting columns.
assert elevation_grid_array.shape == (2, 2)

# Summarize key results without printing large arrays.
print("Settlement list type:", type(settlement_list).__name__)
print("Settlement array type:", type(settlement_array).__name__)
print("Array shape:", settlement_array.shape)
print("Maximum settlement mm:", settlement_array.max())
print("Largest change mm:", settlement_changes.max())
print("First reading inches:", round(settlement_inches[0], 3))
print("Elevation grid shape:", elevation_grid_array.shape)
print("Mean elevation meters:", round(elevation_grid_array.mean(), 2))

```

### **3.2. DataFrame Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1783552892" width="250">



>* DataFrames organize raw data into labeled tables
>* Tables support inspection, filtering, and quality checks

>* DataFrames add labels to numerical arrays
>* Labels make engineering data easier to analyze

>* Use arrays for fast uniform calculations
>* Use DataFrames for labeled engineering summaries



```
#@title Python Code - DataFrame Conversion

# Convert engineering containers into labeled tables.
# DataFrames add context to numerical arrays.
# Small examples keep outputs easy to inspect.

# Import numerical and tabular analysis libraries.
import numpy as np
import pandas as pd

# Store raw bridge observations in Python containers.
bridge_data = {
    "span_id": ["A1", "A2", "B1", "B2"],
    "rating": [7, 6, 5, 8],
    "crack_width_in": [0.03, 0.08, 0.12, 0.02],
}

# Convert the dictionary into a labeled DataFrame.
bridge_df = pd.DataFrame(bridge_data)

# Create a compact NumPy array of pavement measurements.
pavement_array = np.array([
    [0.0, 22.5, 0.18, 95.0],
    [0.5, 23.1, 0.31, 112.0],
    [1.0, 24.0, 0.27, 105.0],
])

# Validate the array shape before assigning column names.
expected_columns = 4
assert pavement_array.shape[1] == expected_columns

# Convert the array into a DataFrame with engineering labels.
pavement_df = pd.DataFrame(
    pavement_array,
    columns=["mile", "temp_C", "rut_in", "iri_in_per_mile"],
)

# Filter rows using meaningful column names.
high_rut_df = pavement_df[pavement_df["rut_in"] > 0.25]

# Convert one DataFrame column back into a NumPy array.
rut_values = pavement_df["rut_in"].to_numpy()

# Summarize the converted structures with limited output.
print("Bridge DataFrame columns:", list(bridge_df.columns))
print("First bridge record:", bridge_df.iloc[0].to_dict())
print("Pavement DataFrame shape:", pavement_df.shape)
print("High rut mile markers:", high_rut_df["mile"].tolist())
print("Mean rut depth, inches:", round(rut_values.mean(), 3))

```

### **3.3. DataFrame Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Civil Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1783552894" width="250">



>* DataFrames organize raw data into labeled tables
>* Rows and columns clarify engineering records

>* Add labels to numerical array data
>* Choose DataFrames for easier tabular analysis

>* Match rows, columns, labels, and units
>* Plan conversions to avoid interpretation errors



```
#@title Python Code - DataFrame Conversion

# Convert engineering containers into labeled tables.
# DataFrames clarify rows, columns, and units.
# Small examples keep conversion steps visible.

# Import numerical and tabular tools.
import numpy as np
import pandas as pd

# Store field measurements in simple containers.
specimen_ids = ["C1", "C2", "C3"]
ages_days = [7, 14, 28]
strengths_psi = [3100, 4200, 5100]

# Convert dictionary columns into one DataFrame.
concrete_data = {
    "specimen_id": specimen_ids,
    "age_days": ages_days,
    "strength_psi": strengths_psi,
}

# Build a labeled concrete test table.
concrete_df = pd.DataFrame(concrete_data)
print("Concrete DataFrame preview:")
print(concrete_df.head(3).to_string(index=False))

# Create a small NumPy sensor array.
sensor_array = np.array([
    [0.12, 0.18, 0.15],
    [0.10, 0.20, 0.17],
])

# Validate the array shape before labeling.
expected_shape = (2, 3)
assert sensor_array.shape == expected_shape

# Convert array readings into labeled columns.
sensor_df = pd.DataFrame(
    sensor_array,
    columns=["floor_1_g", "floor_2_g", "floor_3_g"],
)

# Add row labels representing time steps.
sensor_df.index = ["t0", "t1"]
print("\nSensor DataFrame preview:")
print(sensor_df.to_string())

# Convert selected DataFrame values back to NumPy.
strength_array = concrete_df[["strength_psi"]].to_numpy()
print("\nStrength array shape:", strength_array.shape)
print("Mean strength psi:", round(strength_array.mean(), 1))

```
