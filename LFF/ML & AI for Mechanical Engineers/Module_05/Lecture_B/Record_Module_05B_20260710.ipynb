## **1. Similarity Based Classification**

### **1.1. Engineering Case Similarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_01_01.jpg?v=1783667162" width="250">



>* Classify new cases using similar past examples
>* Similar measured features suggest similar categories

>* Represent cases with consistent measurable features
>* Classify using nearest labeled examples

>* Relevant features reveal meaningful engineering similarity
>* Good comparisons depend on quality past cases



```
#@title Python Code - Engineering Case Similarity

# Compare engineering cases using measured features.
# Nearest neighbors vote from similar examples.
# Scaling changes what similarity means.

import math
from collections import Counter

# Store small bearing cases with two features.
cases = [
    {"name": "A", "vibration": 0.18, "temperature": 42, "label": "healthy"},
    {"name": "B", "vibration": 0.22, "temperature": 45, "label": "healthy"},
    {"name": "C", "vibration": 0.75, "temperature": 88, "label": "faulty"},
    {"name": "D", "vibration": 0.68, "temperature": 82, "label": "faulty"},
]

# Define a new bearing inspection case.
new_case = {"vibration": 0.30, "temperature": 78}

# Compute Euclidean distance between two cases.
def distance(case, query, scales):
    vibration_gap = (case["vibration"] - query["vibration"]) / scales["vibration"]
    temperature_gap = (case["temperature"] - query["temperature"]) / scales["temperature"]
    return math.sqrt(vibration_gap**2 + temperature_gap**2)

# Classify by nearest neighbors and majority vote.
def classify(query, scales, k_value):
    scored = []
    for case in cases:
        gap = distance(case, query, scales)
        scored.append((gap, case["name"], case["label"]))
    neighbors = sorted(scored)[:k_value]
    vote_counts = Counter(label for _, _, label in neighbors)
    prediction = vote_counts.most_common(1)[0][0]
    return prediction, neighbors

# Compare unscaled and scaled similarity choices.
unscaled_scales = {"vibration": 1.0, "temperature": 1.0}
scaled_scales = {"vibration": 0.60, "temperature": 50.0}

# Run two classifications for the same query.
raw_prediction, raw_neighbors = classify(new_case, unscaled_scales, 3)
scaled_prediction, scaled_neighbors = classify(new_case, scaled_scales, 3)

# Print compact teaching results.
print("New bearing: vibration=0.30 in/s, temperature=78 C")
print("Unscaled k=3 prediction:", raw_prediction)
print("Unscaled nearest labels:", [item[2] for item in raw_neighbors])
print("Scaled k=3 prediction:", scaled_prediction)
print("Scaled nearest labels:", [item[2] for item in scaled_neighbors])
print("Lesson: similarity depends on meaningful features and scaling.")

```

### **1.2. Engineering Case Similarity**

```
#@title Python Code - Engineering Case Similarity

# Similarity can classify new engineering cases.
# Nearest neighbors compare measured feature values.
# This example uses manual distance voting.

import math
from collections import Counter

# Store labeled historical machine cases.
cases = [
    ("A", 72.0, 0.18, "normal"),
    ("B", 75.0, 0.22, "normal"),

    ("C", 88.0, 0.55, "inspect"),
    ("D", 91.0, 0.62, "inspect"),
    ("E", 105.0, 1.10, "fail"),

    ("F", 110.0, 1.25, "fail"),
]

# Define one new unlabeled machine reading.
new_case = (90.0, 0.58)
k_value = 3

# Validate the small reference library.
assert len(cases) >= k_value
assert len(new_case) == 2

# Compute Euclidean distance manually.
def distance(case, new_case):
    temp_gap = case[1] - new_case[0]
    vib_gap = case[2] - new_case[1]

    squared_sum = temp_gap**2 + vib_gap**2
    return math.sqrt(squared_sum)

# Rank cases from most similar onward.
distances = []
for case in cases:
    gap = distance(case, new_case)
    distances.append((gap, case[0], case[3]))

# Select the k closest engineering cases.
distances.sort(key=lambda item: item[0])
neighbors = distances[:k_value]
labels = [item[2] for item in neighbors]

# Vote using the neighbor labels.
votes = Counter(labels)
prediction = votes.most_common(1)[0][0]

# Show compact teaching results.
print("New reading: 90 C, 0.58 mm/s vibration")
print("Nearest cases:", neighbors)
print("Neighbor labels:", labels)
print("Majority vote prediction:", prediction)
print("Interpretation: similar past machines guide classification.")

```

### **1.3. Engineering Case Similarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_01_03.jpg?v=1783667164" width="250">



>* Classify new cases using similar past examples
>* KNN structures engineering judgment by analogy

>* Features define similarity between engineering cases
>* Similar measurements suggest similar classifications

```
#@title Python Code - Engineering Case Similarity

# Compare engineering cases using measured features.
# Nearest neighbors classify by practical similarity.
# This example uses manual distance voting.

import math
from collections import Counter

# Store known motor cases with labels.
cases = [
    {"name": "Motor A", "temp_c": 62, "vibration_mm_s": 1.8, "label": "normal"},
    {"name": "Motor B", "temp_c": 68, "vibration_mm_s": 2.2, "label": "normal"},

    {"name": "Motor C", "temp_c": 91, "vibration_mm_s": 6.8, "label": "failure risk"},
    {"name": "Motor D", "temp_c": 88, "vibration_mm_s": 5.9, "label": "failure risk"},
]

# Describe the new engineering case.
new_case = {"temp_c": 84, "vibration_mm_s": 5.4}

# Validate the tiny teaching dataset.
assert len(cases) >= 3 and len(new_case) == 2

# Compute Euclidean distance manually.
def distance(case, target):
    temp_gap = case["temp_c"] - target["temp_c"]
    vib_gap = case["vibration_mm_s"] - target["vibration_mm_s"]

    squared_sum = temp_gap ** 2 + vib_gap ** 2
    return math.sqrt(squared_sum)

# Attach distances to known cases.
distances = []
for case in cases:
    gap = distance(case, new_case)

    distances.append((gap, case["name"], case["label"]))

# Select the three nearest neighbors.
k = 3
nearest = sorted(distances)[:k]

# Vote using the neighbor labels.
labels = [item[2] for item in nearest]
vote_counts = Counter(labels)
prediction = vote_counts.most_common(1)[0][0]

# Print a compact teaching summary.
print("New motor: 84 C, 5.4 mm/s vibration")
print("Nearest known cases:")

for gap, name, label in nearest:
    print(f"{name}: distance {gap:.2f}, label {label}")

print(f"Majority vote with k={k}: {prediction}")
print("KNN classifies by similarity to past engineering cases.")

```

## **2. Manual KNN Implementation**

### **2.1. Distance Calculation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_02_01.jpg?v=1783667153" width="250">



>* Measure similarity using feature-based distances
>* Rank known cases from nearest to farthest

>* Compare identical features in consistent order
>* Combine differences into one distance value

>* Feature scales can distort distance comparisons
>* Scaling helps choose fair nearest neighbors



```
#@title Python Code - Distance Calculation

# Manual KNN starts with distance calculation.
# Similar engineering cases should be nearby.
# Scaling can change measured similarity strongly.

import math
import pandas as pd

# Store tiny labeled component history.
known_cases = pd.DataFrame(
    {
        "case": ["A", "B", "C", "D"],
        "thickness_mm": [10.0, 10.4, 13.2, 12.8],
        "weight_lb": [2.1, 2.3, 3.8, 3.6],
        "roughness_um": [1.2, 1.4, 3.1, 2.9],
        "label": ["acceptable", "acceptable", "defective", "defective"],
    }
)

# Define one new unlabeled component.
new_case = {
    "thickness_mm": 10.6,
    "weight_lb": 2.4,
    "roughness_um": 1.5,
}

# Choose matching feature order carefully.
features = ["thickness_mm", "weight_lb", "roughness_um"]
assert set(features).issubset(known_cases.columns)

# Compute Euclidean distance manually.
def euclidean_distance(row, new_point, feature_names):
    squared_differences = []
    for feature in feature_names:
        difference = row[feature] - new_point[feature]
        squared_differences.append(difference ** 2)

    return math.sqrt(sum(squared_differences))

# Add one distance per known case.
known_cases["distance"] = known_cases.apply(
    lambda row: euclidean_distance(row, new_case, features),
    axis=1,
)

# Sort cases from nearest to farthest.
ranked_cases = known_cases.sort_values("distance")
nearest_columns = ["case", "label", "distance"]
nearest_cases = ranked_cases[nearest_columns].head(3)

# Print compact teaching results.
print("New component features:", new_case)
print("Nearest known cases by manual distance:")
print(nearest_cases.to_string(index=False))
print("Smallest distance means most similar case.")

```

### **2.2. Neighbor Selection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_02_02.jpg?v=1783667158" width="250">



>* Choose the k closest known cases
>* Nearby cases guide the classification decision

>* Sort cases by increasing distance
>* Choose closest records for voting

>* Handle ties and duplicates consistently
>* Choose k to balance noise and locality



```
#@title Python Code - Neighbor Selection

# Manual KNN selects nearby engineering cases.
# Distances are sorted before voting.
# Small k changes the local neighborhood.

# Import only small built-in tools.
from math import sqrt

# Store known motor cases.
cases = [
    ("A", 1.2, 42, "normal"),
    ("B", 1.5, 45, "normal"),

    ("C", 3.8, 80, "fault"),
    ("D", 4.1, 78, "fault"),
    ("E", 2.0, 50, "normal"),

    ("F", 3.2, 70, "fault"),
]

# Define the new motor measurement.
new_case = (2.7, 62)

# Validate the tiny dataset size.
assert len(cases) >= 5, "Need enough known cases."

# Compute distance for every known case.
distances = []
for name, vibration, temperature, label in cases:
    dv = vibration - new_case[0]
    dt = temperature - new_case[1]

    distance = sqrt(dv * dv + dt * dt)
    distances.append((distance, name, label))

# Sort from closest to farthest.
sorted_cases = sorted(distances, key=lambda row: row[0])

# Select neighbors for two k values.
for k in [3, 5]:
    neighbors = sorted_cases[:k]
    labels = [row[2] for row in neighbors]

    normal_votes = labels.count("normal")
    fault_votes = labels.count("fault")
    prediction = "fault" if fault_votes > normal_votes else "normal"

    names = [row[1] for row in neighbors]
    print(f"k={k}: neighbors {names}, prediction={prediction}")

# Show the closest case distance.
closest = sorted_cases[0]
print(f"Closest case: {closest[1]} at distance {closest[0]:.2f}")

```

### **2.3. Majority Voting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_02_03.jpg?v=1783667155" width="250">



>* Classify using most common neighbor label
>* Count nearby labels for a local decision

>* Count selected neighbors’ labels carefully
>* Classify using only nearest relevant examples

>* Plan tie-breaking rules before voting
>* Close votes can change classifications



```
#@title Python Code - Majority Voting

# Manual KNN voting uses nearby labeled cases.
# This example classifies one machine condition.
# Counts decide the final predicted class.

from collections import Counter
import math

# Store tiny engineering cases with two features.
cases = [
    (2.0, 35.0, "normal"),
    (2.4, 38.0, "normal"),

    (6.5, 80.0, "bearing fault"),
    (6.8, 78.0, "bearing fault"),
    (3.0, 90.0, "overheating"),

    (3.2, 88.0, "overheating"),
]

# Define the unknown vibration and temperature case.
new_case = (2.7, 40.0)
k_value = 3

# Check that k is valid for this dataset.
if not 1 <= k_value <= len(cases):
    raise ValueError("k must fit available cases")

# Compute Euclidean distance to every known case.
distances = []
for vibration, temperature, label in cases:
    distance = math.sqrt((vibration - new_case[0]) ** 2 + (temperature - new_case[1]) ** 2)

    distances.append((distance, label))

# Select the k closest labeled neighbors.
distances.sort(key=lambda item: item[0])
neighbors = distances[:k_value]

# Count labels among selected neighbors.
votes = Counter(label for distance, label in neighbors)
winner, count = votes.most_common(1)[0]

# Show only compact teaching output.
print("Nearest neighbor labels:", [label for distance, label in neighbors])
print("Vote counts:", dict(votes))
print("Predicted class:", winner)
print("Majority size:", count, "of", k_value)

```

### **3.1. Feature Scaling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_03_01.jpg?v=1783667166" width="250">



>* Large-scale features can dominate KNN distances
>* Scaling makes neighbor comparisons more balanced

>* Unequal scales distort KNN neighborhoods
>* Scaling can change classification outcomes

>* KNN predictions depend directly on scaled distances
>* Fit scaling on training data consistently



```
#@title Python Code - Feature Scaling

# KNN distances depend strongly on feature scales.
# This example compares raw and scaled neighbors.
# We classify a new mechanical component manually.

import math
import numpy as np
import matplotlib.pyplot as plt

# Store roughness, temperature, and quality labels.
components = np.array([
    [0.4, 820.0], [0.7, 780.0],
    [4.2, 120.0], [4.6, 160.0],
    [0.9, 840.0], [4.8, 140.0]
])

labels = np.array([
    "acceptable", "acceptable", "defective",
    "defective", "acceptable", "defective"
])

# Define one new component for classification.
new_part = np.array([4.4, 760.0])
k_value = 3

# Check the tiny dataset before calculations.
assert components.shape == (6, 2)
assert labels.shape[0] == components.shape[0]

# Compute Euclidean distances without scaling.
raw_distances = np.sqrt(
    np.sum((components - new_part) ** 2, axis=1)
)

# Fit min-max scaling using training components.
mins = components.min(axis=0)
ranges = components.max(axis=0) - mins

# Prevent division by zero during scaling.
ranges = np.where(ranges == 0, 1, ranges)
scaled_components = (components - mins) / ranges

# Apply the same scaling to the new part.
scaled_new_part = (new_part - mins) / ranges
scaled_distances = np.sqrt(
    np.sum((scaled_components - scaled_new_part) ** 2, axis=1)
)

# Select nearest labels and vote manually.
def knn_vote(distances, labels, k):
    order = np.argsort(distances)[:k]
    chosen = labels[order]
    counts = {name: int(np.sum(chosen == name)) for name in set(labels)}
    return max(counts, key=counts.get), chosen

# Compare raw and scaled KNN predictions.
raw_prediction, raw_neighbors = knn_vote(raw_distances, labels, k_value)
scaled_prediction, scaled_neighbors = knn_vote(scaled_distances, labels, k_value)

# Print a compact teaching summary.
print("New part: roughness 4.4 micrometers, temperature 760 C")
print(f"Raw nearest labels: {list(raw_neighbors)}")
print(f"Raw KNN prediction: {raw_prediction}")
print(f"Scaled nearest labels: {list(scaled_neighbors)}")
print(f"Scaled KNN prediction: {scaled_prediction}")
print("Scaling changes distance balance between roughness and temperature.")

# Plot the scaled neighborhood for visual intuition.
colors = ["tab:blue" if x == "acceptable" else "tab:red" for x in labels]
plt.figure(figsize=(6, 4))
plt.scatter(scaled_components[:, 0], scaled_components[:, 1], c=colors, s=80)

# Mark the new component after scaling.
plt.scatter(scaled_new_part[0], scaled_new_part[1], c="black", marker="X", s=140)
plt.xlabel("Scaled surface roughness")
plt.ylabel("Scaled operating temperature")

# Add a concise title and grid.
plt.title("KNN after feature scaling")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.2. Feature Scaling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_03_02.jpg?v=1783667168" width="250">



>* Large-scale features can dominate KNN distances
>* Scaling makes similarity reflect meaningful feature balance

>* Large-scale features can distort distance comparisons
>* Scaling makes KNN similarity more balanced

>* Scaling can change selected nearest neighbors
>* Check features and real-world similarity



```
#@title Python Code - Feature Scaling

# KNN distances depend strongly on feature scales.
# This example compares raw and scaled neighbors.
# Small engineering data keeps calculations transparent.

import math
import numpy as np
import matplotlib.pyplot as plt

# Store roughness, temperature, and quality labels.
roughness = np.array([0.4, 0.7, 1.0, 3.8, 4.2, 4.6])
temperature = np.array([820, 760, 700, 120, 180, 240])
labels = np.array(["acceptable", "acceptable", "acceptable", "defective", "defective", "defective"])

# Define one new component for classification.
new_case = np.array([3.6, 650.0])
features = np.column_stack((roughness, temperature))
assert features.shape == (6, 2)

# Compute Euclidean distances for every known case.
def distances_to_case(data, case):
    differences = data - case
    squared = differences * differences
    return np.sqrt(squared.sum(axis=1))

# Select nearest labels and vote manually.
def knn_vote(data, case, labels, k):
    distances = distances_to_case(data, case)
    order = np.argsort(distances)[:k]
    chosen = labels[order]
    return max(set(chosen), key=list(chosen).count), order

# Scale each feature using training data statistics.
means = features.mean(axis=0)
stds = features.std(axis=0)
scaled_features = (features - means) / stds

# Scale the new case using identical statistics.
scaled_case = (new_case - means) / stds
raw_prediction, raw_neighbors = knn_vote(features, new_case, labels, 3)
scaled_prediction, scaled_neighbors = knn_vote(scaled_features, scaled_case, labels, 3)

# Print concise results for comparison.
print("Raw k=3 prediction:", raw_prediction)
print("Raw neighbor labels:", list(labels[raw_neighbors]))
print("Scaled k=3 prediction:", scaled_prediction)
print("Scaled neighbor labels:", list(labels[scaled_neighbors]))

# Plot raw feature space for visual intuition.
colors = np.where(labels == "acceptable", "tab:blue", "tab:red")
plt.figure(figsize=(6, 4))
plt.scatter(roughness, temperature, c=colors, s=80)

# Mark the new component clearly.
plt.scatter(new_case[0], new_case[1], c="black", marker="X", s=120)
plt.xlabel("Surface roughness, micrometers")
plt.ylabel("Operating temperature, Celsius")
plt.title("Raw units can dominate KNN distance")

# Add a compact legend manually.
plt.scatter([], [], c="tab:blue", label="acceptable")
plt.scatter([], [], c="tab:red", label="defective")
plt.scatter([], [], c="black", marker="X", label="new case")
plt.legend(loc="best")

plt.tight_layout()
plt.show()

```

### **3.3. Choosing k**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_05/Lecture_B/image_03_03.jpg?v=1783667170" width="250">



>* Small k makes decisions highly local
>* Outliers can cause unstable predictions

>* Larger k smooths predictions and reduces noise
>* Too large k ignores local patterns

>* Choose k using validation performance
>* Balance nearby cases with broader patterns



```
#@title Python Code - Choosing k

# Explore how k changes KNN behavior.
# Use tiny engineering measurements for clarity.
# Compare local sensitivity with smoother voting.

import math
import matplotlib.pyplot as plt
from collections import Counter

# Store vibration, temperature, and condition labels.
cases = [
    (1.0, 60.0, "normal"),
    (1.2, 62.0, "normal"),
    (1.4, 63.0, "normal"),

    (2.8, 78.0, "fail"),
    (3.0, 80.0, "fail"),
    (3.2, 82.0, "fail"),
]

# Add one noisy nearby failure example.
cases.append((1.55, 64.0, "fail"))

# Define a new component measurement.
new_case = (1.50, 64.0)

# Confirm the tiny dataset is usable.
assert len(cases) >= 3 and len(new_case) == 2

# Compute Euclidean distance manually.
def distance(point_a, point_b):
    vib_gap = point_a[0] - point_b[0]
    temp_gap = point_a[1] - point_b[1]
    return math.sqrt(vib_gap**2 + temp_gap**2)

# Predict using nearest neighbors and majority vote.
def knn_predict(k_value):
    distances = []
    for vib, temp, label in cases:
        gap = distance(new_case, (vib, temp))
        distances.append((gap, label))

    nearest = sorted(distances)[:k_value]
    votes = Counter(label for gap, label in nearest)
    return votes.most_common(1)[0][0], votes

# Try several odd k values.
k_values = [1, 3, 5, 7]
results = [knn_predict(k_value) for k_value in k_values]

# Print compact results for comparison.
print("KNN predictions for one new component:")
for k_value, result in zip(k_values, results):
    print(f"k={k_value}: prediction={result[0]}, votes={dict(result[1])}")

# Separate points by class for plotting.
normal_points = [(v, t) for v, t, y in cases if y == "normal"]
fail_points = [(v, t) for v, t, y in cases if y == "fail"]

# Create one plot showing the engineering cases.
plt.figure(figsize=(6, 4))
plt.scatter(*zip(*normal_points), label="normal", s=80)
plt.scatter(*zip(*fail_points), label="fail", s=80)

# Mark the new component clearly.
plt.scatter(new_case[0], new_case[1], marker="*", s=220, label="new case")
plt.xlabel("Vibration amplitude, inches per second")
plt.ylabel("Temperature, degrees Celsius")

# Add a short title and legend.
plt.title("Choosing k changes local versus smoother voting")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```
