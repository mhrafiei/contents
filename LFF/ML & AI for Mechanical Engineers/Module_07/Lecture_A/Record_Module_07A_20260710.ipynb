## **1. Layer Concepts**

### **1.1. Weights**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_01_01.jpg?v=1783670775" width="250">



>* Weights control connection influence between neurons
>* They shape inputs by learned importance

>* Training adjusts weights to improve predictions
>* Knowledge is distributed across many weights

>* Weight sign and size shape influence
>* Single weights depend on network context



```
#@title Python Code - Weights

# Weights act like adjustable engineering gains.
# Signs and magnitudes change signal influence.
# NumPy shows one forward weighted sum.

import numpy as np

# Define three normalized engineering input signals.
inputs = np.array([0.8, 0.3, 0.6])

# Define connection weights for one receiving neuron.
weights = np.array([1.5, -0.7, 0.05])

# Validate matching shapes before multiplying signals.
assert inputs.shape == weights.shape

# Multiply each input by its connection weight.
weighted_inputs = inputs * weights

# Combine weighted inputs into one neuron signal.
neuron_signal = np.sum(weighted_inputs)

# Show compact results without printing large arrays.
print("Inputs: floor area, age, distance")
print("Weights: strong positive, negative, near zero")
print("Weighted contributions:", np.round(weighted_inputs, 3))
print("Neuron signal before bias:", round(float(neuron_signal), 3))

# Compare how changing one weight changes influence.
weights_changed = np.array([0.2, -0.7, 0.05])

# Recompute the same neuron with weaker area influence.
changed_signal = np.sum(inputs * weights_changed)

# Print the engineering interpretation briefly.
print("After reducing area weight:", round(float(changed_signal), 3))
print("A larger weight acts like a larger gain.")

```

### **1.2. Hidden Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_01_02.jpg?v=1783670777" width="250">



>* Intermediate neurons transform signals between layers
>* Hidden layers reveal useful internal patterns

>* Layers build simple features into concepts
>* Training learns useful intermediate representations

>* More hidden capacity learns more complex patterns
>* Larger networks need careful practical tradeoffs



```
#@title Python Code - Hidden Layers

# Hidden layers transform engineering measurements.
# Each neuron combines weighted input signals.
# Activations pass useful transformed signals onward.

# Import NumPy for compact array calculations.
import numpy as np

# Use fixed values for repeatable results.
np.set_printoptions(precision=3, suppress=True)

# Two sensor inputs describe one machine state.
inputs = np.array([[80.0, 0.35]])

# Scale temperature and vibration into comparable ranges.
scaled_inputs = np.array([[inputs[0, 0] / 100.0, inputs[0, 1]]])

# Three hidden neurons form intermediate engineering features.
hidden_weights = np.array([[0.8, -0.4, 0.3], [0.6, 0.9, -0.7]])

# Biases shift each hidden neuron's activation threshold.
hidden_biases = np.array([[0.05, -0.10, 0.20]])

# One output neuron combines hidden layer features.
output_weights = np.array([[1.2], [-0.8], [0.5]])

# Output bias shifts the final risk score.
output_bias = np.array([[0.10]])

# Confirm matrix shapes match the network connections.
assert scaled_inputs.shape[1] == hidden_weights.shape[0]

# Compute hidden layer weighted sums.
hidden_sums = scaled_inputs @ hidden_weights + hidden_biases

# ReLU keeps positive signals and blocks negative ones.
hidden_activations = np.maximum(0.0, hidden_sums)

# Confirm hidden outputs connect to output weights.
assert hidden_activations.shape[1] == output_weights.shape[0]

# Compute final regression-style risk score.
risk_score = hidden_activations @ output_weights + output_bias

# Convert score into binary classification probability.
probability = 1.0 / (1.0 + np.exp(-risk_score))

# Print compact teaching results only.
print("Inputs: temperature 80 F, vibration 0.35 in/s")

# Show hidden layer as intermediate features.
print("Hidden activations:", hidden_activations.round(3).tolist())

# Show two suitable output formats.
print("Regression output risk score:", round(float(risk_score[0, 0]), 3))

# Show binary classification output format.
print("Binary classification probability:", round(float(probability[0, 0]), 3))

```

### **1.3. Activation Outputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_01_03.jpg?v=1783670780" width="250">



>* Neurons turn weighted inputs into outgoing signals
>* Activations enable networks to model complex patterns

>* Layer outputs feed the next layer
>* Learned signals encode useful task features

>* Activation functions shape each neuron’s signal
>* Well-shaped outputs help layers learn and communicate



```
#@title Python Code - Activation Outputs

# Activation outputs are processed neuron signals.
# This example uses a tiny network.
# NumPy keeps the forward pass transparent.

import numpy as np

# Define one engineering input sample.
inputs = np.array([[72.0, 0.35]])

# Validate the input shape before multiplying.
assert inputs.shape == (1, 2)

# Store weights for three hidden neurons.
weights_hidden = np.array([[0.08, -1.20, 0.50],
                           [2.00, 0.70, -0.40]])

# Store bias offsets for hidden neurons.
bias_hidden = np.array([[0.10, -0.20, 0.05]])

# Combine inputs, weights, and biases.
raw_hidden = inputs @ weights_hidden + bias_hidden

# ReLU passes positive signals forward.
activation_hidden = np.maximum(0.0, raw_hidden)

# Store weights for one output neuron.
weights_output = np.array([[0.60], [-0.30], [0.80]])

# Store the output neuron bias.
bias_output = np.array([[0.15]])

# Combine hidden activations into output.
raw_output = activation_hidden @ weights_output + bias_output

# Sigmoid compresses output into probability.
activation_output = 1.0 / (1.0 + np.exp(-raw_output))

# Print compact teaching results only.
print("Input: 72 degF, vibration 0.35 in/s")
print("Hidden raw signals:", np.round(raw_hidden, 3))
print("Hidden activation outputs:", np.round(activation_hidden, 3))
print("Output activation probability:", np.round(activation_output, 3))
print("Meaning: activation outputs are signals passed forward.")

```

## **2. Forward Pass**

### **2.1. NumPy Array Setup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_02_01.jpg?v=1783670761" width="250">



>* Organize network data in shaped NumPy arrays
>* Rows hold examples; columns hold features

>* Rows store observations; columns store features
>* Weights and biases match layer neurons

>* Keep array shapes consistent across batch sizes
>* Match dimensions to examples, features, neurons



```
#@title Python Code - NumPy Array Setup

# Set up arrays for neural network inputs.
# Shapes connect engineering data to neuron calculations.
# This example prepares one small forward pass.

import numpy as np

# Use fixed values for repeatable classroom results.
np.set_printoptions(precision=3, suppress=True)

# Rows are houses, columns are measured features.
X = np.array([
    [1200.0, 3.0, 12.0, 4.5],
    [850.0, 2.0, 30.0, 8.0],
])

# Scale features to comparable engineering magnitudes.
scale = np.array([1000.0, 5.0, 50.0, 10.0])
X_scaled = X / scale

# Each column represents one hidden layer neuron.
W1 = np.array([
    [0.8, -0.4, 0.2],
    [0.5, 0.3, -0.6],
    [-0.2, 0.7, 0.4],
    [-0.3, 0.2, 0.9],
])

# One bias value shifts each hidden neuron.
b1 = np.array([0.1, -0.2, 0.05])

# Validate dimensions before matrix multiplication.
assert X_scaled.shape[1] == W1.shape[0]
assert b1.shape[0] == W1.shape[1]

# Compute hidden pre-activation values for every house.
Z1 = X_scaled @ W1 + b1

# ReLU keeps positive signals and clips negatives.
A1 = np.maximum(0.0, Z1)

# Print compact shape information, not large arrays.
print("Input X shape:", X_scaled.shape)
print("Hidden weights W1 shape:", W1.shape)
print("Hidden bias b1 shape:", b1.shape)
print("Hidden activation A1 shape:", A1.shape)
print("First hidden activation row:", A1[0])

```

### **2.2. Matrix Operations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_02_02.jpg?v=1783670763" width="250">



>* Matrix multiplication computes many neuron responses together
>* Inputs and weights produce raw layer outputs

>* Match array shapes to network connections
>* Weights produce multiple neuron responses efficiently

>* Biases shift neuron responses before activation
>* Matrix operations enable efficient forward passes



```
#@title Python Code - Matrix Operations

# Matrix operations power neural network layers.
# Shapes must match connected layer sizes.
# This example uses tiny engineering data.

import numpy as np

# Create two house examples with four features.
inputs = np.array([
    [1800.0, 3.0, 12.0, 4.5],
    [950.0, 2.0, 35.0, 1.2]
])

# Scale features for stable demonstration values.
scaled_inputs = inputs / np.array([2000.0, 5.0, 50.0, 10.0])

# Define weights from four inputs to three neurons.
weights = np.array([
    [0.8, -0.4, 0.2],
    [0.5, 0.3, -0.6],
    [-0.7, 0.9, 0.4],
    [0.2, -0.5, 0.7]
])

# Define one bias value for each neuron.
biases = np.array([0.10, -0.20, 0.05])

# Check that input columns match weight rows.
assert scaled_inputs.shape[1] == weights.shape[0]

# Compute raw neuron responses using matrix multiplication.
raw_hidden = scaled_inputs @ weights + biases

# Apply ReLU activation to keep positive responses.
hidden_outputs = np.maximum(0.0, raw_hidden)

# Print compact shape information for debugging.
print("Input shape:", scaled_inputs.shape)
print("Weight shape:", weights.shape)
print("Bias shape:", biases.shape)
print("Hidden output shape:", hidden_outputs.shape)

# Print rounded outputs without large arrays.
print("Raw hidden responses:", np.round(raw_hidden, 3))
print("Activated hidden outputs:", np.round(hidden_outputs, 3))

# Summarize the engineering interpretation briefly.
print("Each column of weights defines one hidden neuron.")

```

### **2.3. NumPy Forward Pass**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_02_03.jpg?v=1783670765" width="250">



>* Inputs flow through layers to predictions
>* NumPy processes batches efficiently with arrays

>* Track array shapes through each layer
>* Activations turn inputs into useful representations

>* Data flows through hidden and output layers
>* Forward pass predicts without updating parameters



```
#@title Python Code - NumPy Forward Pass

# Demonstrate a small NumPy forward pass.
# Use engineering features and clear shapes.
# Compare regression and classification outputs.

# NumPy is available in Google Colab.
import numpy as np

# Keep printed numbers readable for beginners.
np.set_printoptions(precision=3, suppress=True)

# Create three house examples with engineering-style features.
X = np.array([
    [1200.0, 3.0, 8.0],
    [1800.0, 4.0, 3.0],
    [900.0, 2.0, 12.0],
])

# Scale features to similar numerical ranges.
X_scaled = X / np.array([2000.0, 5.0, 20.0])

# Define weights from three inputs to four hidden neurons.
W1 = np.array([
    [0.8, -0.4, 0.3, 0.2],
    [0.5, 0.7, -0.6, 0.1],
    [-0.3, 0.2, 0.4, 0.9],
])

# Define one bias value for each hidden neuron.
b1 = np.array([0.1, -0.2, 0.05, 0.0])

# Define output weights for regression and classification.
W2 = np.array([[0.6], [-0.5], [0.4], [0.8]])
b2 = np.array([0.15])

# Define ReLU activation for hidden layer outputs.
def relu(values):
    return np.maximum(0.0, values)

# Define sigmoid activation for binary classification output.
def sigmoid(values):
    return 1.0 / (1.0 + np.exp(-values))

# Validate shapes before matrix multiplication.
assert X_scaled.shape == (3, 3)
assert W1.shape == (3, 4)
assert b1.shape == (4,)
assert W2.shape == (4, 1)

# Compute hidden layer weighted sums and activations.
hidden_linear = X_scaled @ W1 + b1
hidden_activation = relu(hidden_linear)

# Compute final raw output from hidden activations.
output_linear = hidden_activation @ W2 + b2

# Interpret raw output as a regression prediction.
price_prediction = 100000.0 + 400000.0 * output_linear

# Interpret sigmoid output as binary classification probability.
failure_probability = sigmoid(output_linear)

# Print compact teaching results under fifteen lines.
print("Input shape:", X_scaled.shape)
print("Hidden activation shape:", hidden_activation.shape)
print("Output shape:", output_linear.shape)
print("Hidden activations:\n", hidden_activation)
print("Regression prices, dollars:", price_prediction.ravel().round(0))
print("Binary probabilities:", failure_probability.ravel().round(3))

```

## **3. Output Format Choices**

### **3.1. Regression Outputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_03_01.jpg?v=1783670748" width="250">



>* Regression predicts continuous numerical values
>* Outputs should match target measurements

>* Match output count to target quantities
>* Scale targets for stable learning

>* Match output range to real quantities
>* Aligned outputs support training and decisions



```
#@title Python Code - Regression Outputs

# Regression outputs represent continuous engineering quantities.
# Output shape should match target measurements.
# Scaling can make learning numerically stable.

import numpy as np

# Create one sample with three engineering features.
features = np.array([[0.8, 0.3, 0.6]])

# Define weights for two regression outputs.
weights = np.array([[12.0, 4.0], [5.0, 2.0], [8.0, 3.0]])

# Biases shift each predicted engineering quantity.
biases = np.array([20.0, 1.5])

# Validate matrix shapes before the forward pass.
assert features.shape[1] == weights.shape[0]

# Compute unconstrained regression outputs using NumPy.
raw_outputs = features @ weights + biases

# Name each output so interpretation stays clear.
delivery_minutes = raw_outputs[0, 0]
energy_kwh = raw_outputs[0, 1]

# Show the output vector shape and meanings.
print(f"Regression output shape: {raw_outputs.shape}")
print(f"Predicted delivery time: {delivery_minutes:.1f} minutes")
print(f"Predicted energy use: {energy_kwh:.1f} kWh")

# Demonstrate a single-output regression format.
single_weight = np.array([[35.0], [10.0], [18.0]])

# Compute one continuous target prediction.
single_output = features @ single_weight + np.array([55.0])

# Print the single numerical prediction.
print(f"Single-output prediction shape: {single_output.shape}")
print(f"Predicted part temperature: {single_output[0, 0]:.1f} °C")

```

### **3.2. Binary Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_03_02.jpg?v=1783670758" width="250">



>* Choose between two possible categories
>* Use one output for class confidence

>* Use 0 and 1 as class labels
>* Keep label meanings consistent and clear

>* Single output estimates class-one confidence
>* Thresholds reflect real error costs



```
#@title Python Code - Binary Classification

# Binary outputs encode two engineering decisions.
# One neuron can represent class confidence.
# Thresholds convert confidence into final labels.

import numpy as np

# Store two sensor features per part.
features = np.array([
    [0.20, 0.10],
    [0.80, 0.70],
    [0.45, 0.55],
])

# Validate the tiny input matrix shape.
assert features.shape == (3, 2)

# Define one output neuron manually.
weights = np.array([3.0, 2.0])
bias = -2.5

# Compute raw scores before activation.
raw_scores = features @ weights + bias

# Convert scores into class-one confidence.
probabilities = 1.0 / (1.0 + np.exp(-raw_scores))

# Choose a decision threshold deliberately.
threshold = 0.60
predicted_labels = (probabilities >= threshold).astype(int)

# Map numeric labels to engineering meanings.
label_names = {0: "acceptable", 1: "defective"}

# Print compact teaching results only.
print("Binary classification output format")
print("Target coding: 0 = acceptable, 1 = defective")
print(f"Decision threshold: {threshold:.2f}")

# Show each part confidence and decision.
for index, probability in enumerate(probabilities, start=1):
    label = predicted_labels[index - 1]
    name = label_names[int(label)]
    print(f"Part {index}: P(defective)={probability:.2f} -> {name}")

# Emphasize that probabilities are not measurements.
print("The single output is confidence for class 1.")

```

### **3.3. Sigmoid Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_07/Lecture_A/image_03_03.jpg?v=1783670750" width="250">



>* Binary decisions use one sigmoid score
>* Scores show confidence between zero and one

>* Convert sigmoid scores using chosen thresholds
>* Adjust cutoffs for application risk

>* Treat sigmoid scores as calibrated estimates
>* Check data quality and decision risks



```
#@title Python Code - Sigmoid Output

# Sigmoid outputs support binary engineering decisions.
# Scores become probabilities between zero and one.
# Thresholds convert scores into final class labels.

import numpy as np
import matplotlib.pyplot as plt

# Define a stable sigmoid activation function.
def sigmoid(raw_scores):
    clipped_scores = np.clip(raw_scores, -50, 50)
    return 1 / (1 + np.exp(-clipped_scores))

# Store simple inspection features for four parts.
features = np.array([
    [0.02, 0.10],
    [0.08, 0.30],

    [0.15, 0.55],
    [0.25, 0.80]
])

# Validate the feature matrix shape.
assert features.shape == (4, 2)

# Choose weights and bias for defect risk.
weights = np.array([8.0, 3.0])
bias = -2.2

# Compute raw network outputs before sigmoid.
raw_scores = features @ weights + bias
probabilities = sigmoid(raw_scores)

# Convert probabilities using two practical thresholds.
standard_labels = probabilities >= 0.50
cautious_labels = probabilities >= 0.30

# Print compact results for engineering interpretation.
print("Sigmoid output example: defect probability")
print("Part | score | p(defect) | label@0.50 | label@0.30")

# Display each part without printing large arrays.
for index, probability in enumerate(probabilities, start=1):
    standard = "defect" if standard_labels[index - 1] else "acceptable"
    cautious = "defect" if cautious_labels[index - 1] else "acceptable"

    print(f"{index:>4} | {raw_scores[index-1]:>5.2f} | {probability:>8.2f} | {standard:>10} | {cautious:>10}")

# Plot sigmoid curve and selected thresholds.
x_values = np.linspace(-6, 6, 200)
y_values = sigmoid(x_values)

# Create one clear teaching plot.
plt.figure(figsize=(7, 4))
plt.plot(x_values, y_values, label="sigmoid(score)")
plt.axhline(0.50, linestyle="--", label="threshold 0.50")

plt.axhline(0.30, linestyle=":", label="threshold 0.30")
plt.xlabel("Raw network score")
plt.ylabel("Sigmoid output probability")
plt.title("Sigmoid Output for Binary Classification")

plt.ylim(-0.05, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

```
