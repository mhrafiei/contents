```
#@title Python Code - Image Arrays

# Build a tiny inspection image.
# Store pixels inside NumPy arrays.
# Summarize image regions with features.

# NumPy and matplotlib are usually preinstalled.
import numpy as np
import matplotlib.pyplot as plt

# Create a bright metal surface image.
image = np.full((8, 8), 180, dtype=np.uint8)
image[3:5, 2:6] = 45

# Validate the image grid shape.
assert image.shape == (8, 8)
assert image.ndim == 2

# Extract simple inspection features.
mean_brightness = float(image.mean())
dark_pixels = int(np.sum(image < 80))

# Compare two physical image regions.
top_half = image[:4, :]
bottom_half = image[4:, :]

# Store region summaries for printing.
region_means = {
    "top_mean": float(top_half.mean()),
    "bottom_mean": float(bottom_half.mean()),
}

# Print compact teaching results.
print("Image shape:", image.shape)
print("Pixel type:", image.dtype)
print("Mean brightness:", round(mean_brightness, 1))
print("Dark defect pixels:", dark_pixels)
print("Region means:", region_means)

# Display the array as an image.
plt.figure(figsize=(4, 4))
plt.imshow(image, cmap="gray", vmin=0, vmax=255)

# Add a clear inspection title.
plt.title("Tiny grayscale inspection image")
plt.colorbar(label="brightness value")
plt.xticks(range(8))

# Label rows and columns clearly.
plt.yticks(range(8))
plt.xlabel("column position")
plt.ylabel("row position")

# Show the single teaching plot.
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Array Shape Basics

# Learn how image arrays store inspection grids.
# Shapes describe rows, columns, and channels.
# Small arrays keep the example readable.

# Import NumPy for compact numerical image arrays.
import numpy as np

# Create one tiny grayscale inspection image.
gray_image = np.array([
    [20, 25, 30, 35],
    [22, 80, 85, 38],
    [24, 82, 90, 40],
    [26, 28, 32, 42],
])

# Create three color channels from grayscale values.
red_channel = gray_image + 10
green_channel = gray_image.copy()
blue_channel = gray_image - 10

# Stack channels into one color image array.
color_image = np.stack(
    [red_channel, green_channel, blue_channel],
    axis=2,
)

# Build a batch containing two grayscale images.
second_gray = gray_image + 5
gray_batch = np.stack([gray_image, second_gray], axis=0)

# Validate expected shapes before printing summaries.
assert gray_image.shape == (4, 4)
assert color_image.shape == (4, 4, 3)
assert gray_batch.shape == (2, 4, 4)

# Print concise shape summaries for learners.
print(f"Grayscale image shape: {gray_image.shape}")
print(f"Color image shape: {color_image.shape}")
print(f"Batch of grayscale images shape: {gray_batch.shape}")
print(f"Rows and columns: {gray_image.shape[0]} by {gray_image.shape[1]}")
print(f"Color channels per pixel: {color_image.shape[2]}")
print(f"Number of images in batch: {gray_batch.shape[0]}")

```

### **1.3. Sensor Time Tables**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_01_03.jpg?v=1783673507" width="250">



>* Rows capture timed sensor observations
>* Time order gives measurements meaning

>* Timestamps order and align sensor readings
>* Tables support filtering, summaries, and comparisons

>* Define each row’s timing meaning
>* Check columns, sampling, and data issues



```
#@title Python Code - Sensor Time Tables

# Sensor tables organize changing engineering measurements.
# Each row represents one recorded moment.
# Columns store time, sensors, and labels.

# Import compact table and numeric tools.
import numpy as np
import pandas as pd

# Create a short regular time sequence.
time_index = pd.date_range(
    start="2026-01-01 08:00:00",
    periods=8,
    freq="1min"
)

# Store simple oven sensor readings.
temperature_c = np.array([22, 35, 49, 63, 76, 82, 84, 83])
vibration_mm_s = np.array([0.8, 0.9, 1.1, 1.4, 1.8, 2.2, 2.5, 2.4])
mode_labels = ["warmup"] * 4 + ["steady"] * 4

# Build one row per recorded moment.
sensor_table = pd.DataFrame({
    "time": time_index,
    "temperature_C": temperature_c,
    "vibration_mm_s": vibration_mm_s,
    "mode": mode_labels
})

# Add beginner friendly time features.
sensor_table["minute"] = sensor_table["time"].dt.minute
sensor_table["temp_change_C"] = sensor_table["temperature_C"].diff().fillna(0)

# Validate the table before summarizing.
expected_rows = 8
assert len(sensor_table) == expected_rows
assert sensor_table["time"].is_monotonic_increasing

# Print only a compact preview.
print("Sensor time table preview:")
print(sensor_table.head(3).to_string(index=False))

# Summarize useful engineering features.
mean_temp = sensor_table["temperature_C"].mean()
max_vibration = sensor_table["vibration_mm_s"].max()
fastest_rise = sensor_table["temp_change_C"].max()

# Print concise feature results.
print(f"Mean temperature: {mean_temp:.1f} C")
print(f"Maximum vibration: {max_vibration:.1f} mm/s")
print(f"Fastest temperature rise: {fastest_rise:.1f} C/min")

# Plot two sensor columns against time.
ax = sensor_table.plot(
    x="time",
    y=["temperature_C", "vibration_mm_s"],
    marker="o",
    figsize=(7, 4)
)

# Label the single teaching plot.
ax.set_title("Small Sensor Time Table")
ax.set_xlabel("Recorded time")
ax.set_ylabel("Sensor value")
ax.grid(True)

# Display the plot in Colab.
import matplotlib.pyplot as plt
plt.tight_layout()
plt.show()

```

### **2.1. Rolling Statistics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_02_01.jpg?v=1783673518" width="250">



>* Summarize recent signal behavior with moving windows
>* Capture local trends and variability over time

>* Window size controls captured time scale
>* Multiple windows reveal patterns for models

>* Spot early changes in signals
>* Use only past data for predictions



```
#@title Python Code - Rolling Statistics

# Rolling statistics summarize recent sensor behavior.
# Small windows reveal fast local changes.
# Larger windows smooth noisy engineering signals.

# Import common data and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic vibration signal.
np.random.seed(7)
seconds = np.arange(0, 60, 1)
trend = 0.02 * seconds

# Add noise and one short disturbance.
noise = np.random.normal(0, 0.08, seconds.size)
spike = np.where((seconds >= 32) & (seconds <= 36), 0.9, 0.0)
vibration = 1.5 + trend + noise + spike

# Store the signal in a table.
data = pd.DataFrame({"second": seconds, "vibration_mm_s": vibration})
assert len(data) == 60

# Compute rolling features using past values.
data["roll_mean_5s"] = data["vibration_mm_s"].rolling(5).mean()
data["roll_std_5s"] = data["vibration_mm_s"].rolling(5).std()
data["roll_max_10s"] = data["vibration_mm_s"].rolling(10).max()

# Show only selected rows for readability.
preview = data.loc[28:38, ["second", "vibration_mm_s", "roll_mean_5s", "roll_std_5s"]]
print("Rolling feature preview around the disturbance:")
print(preview.round(3).to_string(index=False))

# Compare raw signal and rolling average.
plt.figure(figsize=(8, 4))
plt.plot(data["second"], data["vibration_mm_s"], label="raw vibration")
plt.plot(data["second"], data["roll_mean_5s"], label="5-second rolling mean")

# Label the engineering time-series plot.
plt.xlabel("Time (seconds)")
plt.ylabel("Vibration speed (mm/s)")
plt.title("Rolling Statistics Highlight Local Signal Behavior")
plt.legend()

# Display one compact teaching plot.
plt.tight_layout()
plt.show()

```

### **2.2. Peak Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_02_02.jpg?v=1783673520" width="250">



>* Peaks mark important highs and lows
>* Summarize extreme events for modeling

>* Separate true peaks from normal noise
>* Match peak definitions to engineering meaning

>* Peak features make model patterns easier to explain
>* Clean data prevents misleading extreme-value features



```
#@title Python Code - Peak Features

# Peak features summarize important signal events.
# Small examples help connect features physically.
# This script uses synthetic vibration data.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a deterministic time axis.
np.random.seed(7)
time_seconds = np.linspace(0, 10, 101)
base_signal = 0.25 * np.sin(2 * np.pi * time_seconds)

# Add small measurement noise.
noise_signal = np.random.normal(0, 0.06, time_seconds.size)
vibration_g = base_signal + noise_signal
assert vibration_g.size == time_seconds.size

# Insert three visible vibration peaks.
peak_indices = np.array([22, 55, 82])
peak_boosts = np.array([1.10, 0.85, 1.35])
vibration_g[peak_indices] = vibration_g[peak_indices] + peak_boosts

# Store the signal as a table.
sensor_table = pd.DataFrame({"time_s": time_seconds, "vibration_g": vibration_g})
threshold_g = 0.75
above_threshold = sensor_table["vibration_g"] > threshold_g

# Extract beginner friendly peak features.
max_value = sensor_table["vibration_g"].max()
min_value = sensor_table["vibration_g"].min()
peak_count = int(above_threshold.sum())
max_time = sensor_table.loc[sensor_table["vibration_g"].idxmax(), "time_s"]

# Estimate peak strength above typical level.
typical_level = sensor_table["vibration_g"].median()
peak_margin = max_value - typical_level
feature_table = pd.DataFrame({"feature": ["max_g", "min_g", "peak_count", "max_time_s", "peak_margin_g"], "value": [max_value, min_value, peak_count, max_time, peak_margin]})

# Print compact feature results.
print("Peak feature summary for vibration signal:")
print(feature_table.round(3).to_string(index=False))
print("Threshold used:", threshold_g, "g")

# Plot the signal and threshold.
plt.figure(figsize=(8, 4))
plt.plot(sensor_table["time_s"], sensor_table["vibration_g"], marker="o", linewidth=1)
plt.axhline(threshold_g, color="red", linestyle="--", label="peak threshold")

# Highlight readings counted as peaks.
plt.scatter(sensor_table.loc[above_threshold, "time_s"], sensor_table.loc[above_threshold, "vibration_g"], color="red", s=60, label="counted peaks")
plt.xlabel("Time (seconds)")
plt.ylabel("Vibration acceleration (g)")
plt.title("Peak Features From a Sensor Time Series")

# Show one clear teaching plot.
plt.legend()
plt.tight_layout()
plt.show()

```

### **2.3. Trend Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_02_03.jpg?v=1783673522" width="250">



>* Summarize overall signal direction and change rate
>* Connect trends to real machine behavior

>* Compare early and late signal values
>* Describe increases, decreases, drift, or stability

>* Match trends to time scale and context
>* Combine features to capture system evolution



```
#@title Python Code - Trend Features

# Trend features summarize overall signal direction.
# Small sensor examples support engineering intuition.
# NumPy and pandas make features readable.

# Required libraries are already available.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a small deterministic time axis.
time_seconds = np.arange(0, 12, 1)
assert time_seconds.size == 12

# Build three simple vibration signals.
stable_signal = 0.30 + 0.01 * np.sin(time_seconds)
rising_signal = 0.20 + 0.025 * time_seconds
falling_signal = 0.55 - 0.020 * time_seconds

# Store signals in a compact table.
signals = pd.DataFrame({
    "time_s": time_seconds,
    "stable_in_s": stable_signal,
    "rising_in_s": rising_signal,
    "falling_in_s": falling_signal,
})

# Define a beginner friendly trend extractor.
def trend_features(values, times):
    values = np.asarray(values, dtype=float)
    times = np.asarray(times, dtype=float)

    # Validate matching one dimensional inputs.
    assert values.ndim == 1 and times.ndim == 1
    assert values.size == times.size and values.size >= 4

    # Compare early and late average levels.
    early_mean = values[:3].mean()
    late_mean = values[-3:].mean()

    # Estimate slope using a simple line fit.
    slope = np.polyfit(times, values, 1)[0]
    total_change = values[-1] - values[0]

    # Return compact trend feature values.
    return pd.Series({
        "early_mean": early_mean,
        "late_mean": late_mean,
        "total_change": total_change,
        "slope_per_second": slope,
    })

# Extract trend features for each signal.
feature_table = signals.drop(columns="time_s").apply(
    lambda column: trend_features(column, signals["time_s"])
).T

# Add a simple direction label.
feature_table["direction"] = np.where(
    feature_table["slope_per_second"] > 0.005,
    "increasing",
    np.where(feature_table["slope_per_second"] < -0.005, "decreasing", "stable"),
)

# Print a compact teaching summary.
print("Trend features from small vibration signals:")
print(feature_table.round(4).to_string())

# Plot the raw signals for visual comparison.
plt.figure(figsize=(7, 4))
plt.plot(signals["time_s"], signals["stable_in_s"], marker="o", label="stable")
plt.plot(signals["time_s"], signals["rising_in_s"], marker="o", label="rising")
plt.plot(signals["time_s"], signals["falling_in_s"], marker="o", label="falling")

# Label the engineering time series plot.
plt.xlabel("Time (seconds)")
plt.ylabel("Vibration amplitude (inches/second)")
plt.title("Trend features describe overall direction")
plt.legend()

# Display exactly one plot.
plt.tight_layout()
plt.show()

```

## **3. Cross Modal Models**

### **3.1. Shared Evaluation Setup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_03_01.jpg?v=1783673513" width="250">



>* Use identical tasks, labels, splits, and metrics
>* Compare features without hidden testing differences

>* Define one prediction unit clearly
>* Align all data to each case

>* Use identical splits and training-only preprocessing
>* Report shared metrics for fair comparison



```
#@title Python Code - Shared Evaluation Setup

# Compare modalities using one shared split.
# Tiny synthetic inspection data stays readable.
# NumPy and pandas handle all features.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Fix randomness for repeatable lecture results.
rng = np.random.default_rng(42)
n_parts = 80
part_id = np.arange(n_parts)

# Create one target for every part.
wear_mm = rng.normal(0.45, 0.18, n_parts)
heat_c = rng.normal(72.0, 8.0, n_parts)
defect = ((wear_mm + heat_c / 200) > 0.86).astype(int)

# Build tabular measurements for each part.
tabular = pd.DataFrame({"wear_mm": wear_mm, "heat_c": heat_c})
tabular["length_in"] = rng.normal(4.0, 0.08, n_parts)

# Build image-style grids aligned by part.
images = rng.normal(0.25, 0.05, (n_parts, 8, 8))
images += defect[:, None, None] * rng.normal(0.18, 0.03, (n_parts, 8, 8))
image_features = pd.DataFrame({"image_mean": images.mean(axis=(1, 2))})

# Build time signals aligned by part.
time_axis = np.linspace(0, 1, 30)
signals = rng.normal(0.0, 0.08, (n_parts, 30))
signals += defect[:, None] * np.sin(2 * np.pi * 4 * time_axis)
time_features = pd.DataFrame({"signal_rms": np.sqrt((signals ** 2).mean(axis=1))})

# Use one shared train test split.
indices = rng.permutation(n_parts)
train_idx = indices[:56]
test_idx = indices[56:]
assert len(set(train_idx) & set(test_idx)) == 0

# Define a tiny nearest-centroid classifier.
def centroid_predict(train_x, train_y, test_x):
    classes = np.array([0, 1])
    centers = np.vstack([train_x[train_y == c].mean(axis=0) for c in classes])
    distances = ((test_x[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
    return classes[distances.argmin(axis=1)]

# Evaluate each feature set identically.
def evaluate(name, frame):
    x = frame.to_numpy(dtype=float)
    train_x = x[train_idx]
    test_x = x[test_idx]
    train_y = defect[train_idx]
    test_y = defect[test_idx]
    pred = centroid_predict(train_x, train_y, test_x)
    return name, (pred == test_y).mean()

# Compare modalities under shared conditions.
results = [
    evaluate("Tabular", tabular),
    evaluate("Image features", image_features),
    evaluate("Time features", time_features),
]

# Print compact results for discussion.
print("Shared target, split, and metric used for every model.")
for name, acc in results:
    print(f"{name:15s} accuracy: {acc:.2f}")

# Plot one concise comparison chart.
labels = [item[0] for item in results]
scores = [item[1] for item in results]
plt.figure(figsize=(6, 3))
plt.bar(labels, scores, color=["steelblue", "orange", "seagreen"])
plt.ylim(0, 1)
plt.ylabel("Test accuracy")
plt.title("Fair comparison using one shared evaluation setup")
plt.show()

```

### **3.2. Feature Set Comparison**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_03_02.jpg?v=1783673511" width="250">



>* Different data views reveal different evidence
>* Match features to the engineering decision

>* Match feature types to defect causes
>* Compare gains from spatial or temporal evidence

>* Balance accuracy gains with practical costs
>* Choose stable, maintainable feature sets



```
#@title Python Code - Feature Set Comparison

# Compare feature sets for one inspection task.
# Tiny synthetic data keeps execution fast.
# Simple NumPy models avoid extra libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a deterministic random generator.
rng = np.random.default_rng(42)
n_parts = 90

# Create hidden defect severity values.
severity = rng.normal(0.0, 1.0, n_parts)
noise = rng.normal(0.0, 0.35, n_parts)

# Build tabular production measurements.
feed_ipm = 20.0 + 1.8 * severity + noise
thickness_mm = 5.0 - 0.25 * severity + noise

# Build image-style grid features.
bright_spots = 4.0 + 2.5 * severity + rng.normal(0, 1, n_parts)
edge_count = 30.0 + 8.0 * severity + rng.normal(0, 4, n_parts)

# Build time-series signal features.
vibration_peak = 1.0 + 0.9 * severity + rng.normal(0, 0.25, n_parts)
temperature_slope = 0.2 + 0.35 * severity + rng.normal(0, 0.12, n_parts)

# Define the inspection outcome.
defective = (severity + rng.normal(0, 0.45, n_parts)) > 0.25
labels = defective.astype(int)

# Store features in a compact table.
data = pd.DataFrame({
    "feed_ipm": feed_ipm,
    "thickness_mm": thickness_mm,
    "bright_spots": bright_spots,
    "edge_count": edge_count,
    "vibration_peak": vibration_peak,
    "temperature_slope": temperature_slope,
    "defective": labels,
})

# Validate the small teaching dataset.
assert data.shape == (n_parts, 7)
assert set(data["defective"].unique()).issubset({0, 1})

# Define feature groups as different lenses.
feature_sets = {
    "Tabular": ["feed_ipm", "thickness_mm"],
    "Image-style": ["bright_spots", "edge_count"],
    "Time-series": ["vibration_peak", "temperature_slope"],
    "Combined": ["feed_ipm", "thickness_mm", "bright_spots", "edge_count", "vibration_peak", "temperature_slope"],
}

# Split rows without shuffling labels separately.
indices = rng.permutation(n_parts)
train_idx = indices[:60]
test_idx = indices[60:]

# Fit a tiny nearest-centroid classifier.
def centroid_accuracy(columns):
    X_train = data.loc[train_idx, columns].to_numpy()
    X_test = data.loc[test_idx, columns].to_numpy()
    y_train = labels[train_idx]
    y_test = labels[test_idx]

    # Standardize using training statistics only.
    mean = X_train.mean(axis=0)
    scale = X_train.std(axis=0) + 1e-9
    X_train = (X_train - mean) / scale
    X_test = (X_test - mean) / scale

    # Compute one centroid per class.
    c0 = X_train[y_train == 0].mean(axis=0)
    c1 = X_train[y_train == 1].mean(axis=0)
    d0 = np.linalg.norm(X_test - c0, axis=1)
    d1 = np.linalg.norm(X_test - c1, axis=1)

    # Predict the closer centroid.
    predictions = (d1 < d0).astype(int)
    return float((predictions == y_test).mean())

# Evaluate each representation fairly.
results = {name: centroid_accuracy(cols) for name, cols in feature_sets.items()}
summary = pd.Series(results).sort_values(ascending=False)

# Print a compact comparison table.
print("Feature set comparison using one simple classifier:")
for name, score in summary.items():
    print(f"{name:12s} accuracy = {score:.2f}")

# Plot exactly one compact bar chart.
plt.figure(figsize=(7, 4))
plt.bar(summary.index, summary.values, color=["#4C78A8", "#F58518", "#54A24B", "#B279A2"])
plt.ylim(0, 1.05)
plt.ylabel("Test accuracy")
plt.title("Same task, different feature sets")
plt.grid(axis="y", alpha=0.25)
plt.show()

```

### **3.3. Model Comparison**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_08/Lecture_A/image_03_03.jpg?v=1783673516" width="250">



>* Compare which data form gives useful evidence
>* Choose models learning stable, meaningful patterns

>* Compare matching models and feature types
>* Choose practical models when performance is similar

>* Balance accuracy with real-world constraints
>* Use error impacts to guide decisions



```
#@title Python Code - Model Comparison

# Compare modalities using simple engineered features.
# Tiny synthetic data keeps execution fast.
# Accuracy is only one decision clue.

# Import numerical and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make results deterministic for teaching.
rng = np.random.default_rng(42)
n_parts = 90
assert n_parts >= 30

# Create a hidden defect severity signal.
severity = rng.normal(0.0, 1.0, n_parts)
labels = (severity > 0.25).astype(int)
assert labels.shape[0] == n_parts

# Build tabular inspection measurements.
age_months = 24 + 5 * severity + rng.normal(0, 4, n_parts)
temp_c = 70 + 8 * severity + rng.normal(0, 5, n_parts)
vibration_ips = 0.12 + 0.05 * severity + rng.normal(0, 0.04, n_parts)

# Store tabular features in pandas.
tabular = pd.DataFrame({
    "age_months": age_months,
    "temp_c": temp_c,
    "vibration_ips": vibration_ips,
})

# Build image-style grids with localized hot spots.
images = rng.normal(0.25, 0.05, (n_parts, 8, 8))
images[:, 3:5, 3:5] += severity[:, None, None] * 0.18
assert images.shape == (n_parts, 8, 8)

# Build time-series vibration signals.
time = np.linspace(0, 1, 40)
signals = rng.normal(0, 0.05, (n_parts, 40))
signals += (0.25 + 0.12 * severity[:, None]) * np.sin(2 * np.pi * 6 * time)
assert signals.shape == (n_parts, 40)

# Extract beginner-friendly image features.
image_features = pd.DataFrame({
    "mean_intensity": images.mean(axis=(1, 2)),
    "center_patch_mean": images[:, 3:5, 3:5].mean(axis=(1, 2)),
    "max_intensity": images.max(axis=(1, 2)),
})

# Extract beginner-friendly time features.
time_features = pd.DataFrame({
    "signal_mean": signals.mean(axis=1),
    "signal_peak": np.abs(signals).max(axis=1),
    "signal_std": signals.std(axis=1),
})

# Define a tiny nearest-centroid classifier.
def centroid_accuracy(features, target):
    values = np.asarray(features, dtype=float)
    target = np.asarray(target, dtype=int)
    split = int(0.7 * len(target))

    train_x, test_x = values[:split], values[split:]
    train_y, test_y = target[:split], target[split:]
    assert test_x.shape[0] > 0

    mean = train_x.mean(axis=0)
    std = train_x.std(axis=0) + 1e-9
    train_z = (train_x - mean) / std
    test_z = (test_x - mean) / std

    centers = np.vstack([train_z[train_y == 0].mean(axis=0), train_z[train_y == 1].mean(axis=0)])
    distances = ((test_z[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
    predictions = distances.argmin(axis=1)
    return (predictions == test_y).mean()

# Compare the same simple model across modalities.
results = pd.DataFrame({
    "feature_set": ["Tabular", "Image-style", "Time-series"],
    "accuracy": [centroid_accuracy(tabular, labels), centroid_accuracy(image_features, labels), centroid_accuracy(time_features, labels)],
})

# Print a compact comparison table.
print("Nearest-centroid model comparison:")
print(results.to_string(index=False, formatters={"accuracy": "{:.2f}".format}))

# Plot one concise performance comparison.
plt.figure(figsize=(6, 3.5))
plt.bar(results["feature_set"], results["accuracy"], color=["steelblue", "darkorange", "seagreen"])
plt.ylim(0, 1)
plt.ylabel("Test accuracy")
plt.title("Same simple model, different feature sets")
plt.tight_layout()
plt.show()

```
