## **1. Baseline Models**

### **1.1. Regression Baselines**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1783661256" width="250">



>* Simple predictors set minimum performance expectations
>* Baselines test whether complexity adds value

>* Mean, median, and persistence baselines
>* Simple, transparent, and quick predictors

>* Judge whether model gains justify costs
>* Expose data issues and weak models



```
#@title Python Code - Regression Baselines

# Regression baselines set simple performance expectations.
# Engineering predictions need practical comparison points.
# This example compares mean and median baselines.

import numpy as np
import matplotlib.pyplot as plt

# Store small pump energy observations.
train_kwh = np.array([48, 51, 49, 52, 50, 120], dtype=float)
test_kwh = np.array([50, 53, 49, 55], dtype=float)

# Validate arrays before computing metrics.
assert train_kwh.size > 0 and test_kwh.size > 0
assert train_kwh.ndim == 1 and test_kwh.ndim == 1

# Build two simple baseline predictors.
mean_prediction = float(np.mean(train_kwh))
median_prediction = float(np.median(train_kwh))

# Repeat each baseline for every test case.
mean_preds = np.full(test_kwh.shape, mean_prediction)
median_preds = np.full(test_kwh.shape, median_prediction)

# Define mean squared error manually.
def mean_squared_error(actual_values, predicted_values):
    errors = actual_values - predicted_values
    squared_errors = errors ** 2
    return float(np.mean(squared_errors))

# Evaluate both regression baselines.
mean_mse = mean_squared_error(test_kwh, mean_preds)
median_mse = mean_squared_error(test_kwh, median_preds)

# Print compact engineering interpretation.
print(f"Mean baseline prediction: {mean_prediction:.1f} kWh")
print(f"Median baseline prediction: {median_prediction:.1f} kWh")
print(f"Mean baseline MSE: {mean_mse:.2f} kWh^2")
print(f"Median baseline MSE: {median_mse:.2f} kWh^2")

# Choose the lower risk baseline.
better_name = "median" if median_mse < mean_mse else "mean"
print(f"Lower test error baseline: {better_name}")

# Plot actual values and baseline predictions.
days = np.arange(1, test_kwh.size + 1)
plt.figure(figsize=(7, 4))
plt.plot(days, test_kwh, marker="o", label="Actual test kWh")
plt.plot(days, mean_preds, marker="s", label="Mean baseline")
plt.plot(days, median_preds, marker="^", label="Median baseline")
plt.xlabel("Test day")
plt.ylabel("Pump energy use, kWh")
plt.title("Regression baselines for pump energy prediction")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Regression Baselines**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1783661258" width="250">



>* Use simple averages as regression baselines
>* Compare complex models against this starting point

>* Use medians or recent values when appropriate
>* Compare against simple engineering rules

>* Compare model errors against simple baselines
>* Justify complexity through risk and cost benefits



```
#@title Python Code - Regression Baselines

# Regression baselines compare simple prediction strategies.
# Engineering teams need clear reference performance.
# This example predicts pump flow rates.

import numpy as np
import matplotlib.pyplot as plt

# Create small deterministic engineering training data.
train_hours = np.array([10, 20, 30, 40, 50, 60])
train_flow = np.array([98, 101, 99, 103, 250, 102])

# Create small test data with true outcomes.
test_hours = np.array([15, 35, 55, 75])
test_flow = np.array([100, 102, 104, 103])

# Validate matching target and feature lengths.
assert train_hours.size == train_flow.size
assert test_hours.size == test_flow.size

# Build mean and median regression baselines.
mean_prediction = float(np.mean(train_flow))
median_prediction = float(np.median(train_flow))

# Repeat each baseline prediction for every test case.
mean_baseline = np.full(test_flow.shape, mean_prediction)
median_baseline = np.full(test_flow.shape, median_prediction)

# Define manual mean squared error calculation.
def mean_squared_error(actual_values, predicted_values):
    errors = actual_values - predicted_values
    squared_errors = errors ** 2

    return float(np.mean(squared_errors))

# Evaluate both simple baseline strategies.
mean_mse = mean_squared_error(test_flow, mean_baseline)
median_mse = mean_squared_error(test_flow, median_baseline)

# Print compact results for engineering interpretation.
print(f"Mean baseline prediction: {mean_prediction:.1f} gallons per minute")
print(f"Median baseline prediction: {median_prediction:.1f} gallons per minute")
print(f"Mean baseline MSE: {mean_mse:.2f}")
print(f"Median baseline MSE: {median_mse:.2f}")
print("Lower MSE means better baseline decision quality.")

# Plot true test values against baseline predictions.
plt.figure(figsize=(7, 4))
plt.plot(test_hours, test_flow, "o-", label="Actual flow")
plt.plot(test_hours, mean_baseline, "s--", label="Mean baseline")

plt.plot(test_hours, median_baseline, "d--", label="Median baseline")
plt.xlabel("Operating time, hours")
plt.ylabel("Pump flow, gallons per minute")
plt.title("Regression Baselines for Pump Flow")

plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **1.3. Baseline Predictor Types**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1783661260" width="250">



>* Simple baselines set comparison starting points
>* Use mean, median, or majority class

>* Constant and random baselines reveal chance performance
>* Stratified and rule-based baselines add context

>* Match baselines to context and decisions
>* Weak improvement signals deeper modeling problems



```
#@title Python Code - Baseline Predictor Types

# Baselines give simple engineering comparison points.
# We compare regression and classification baselines.
# Metrics show whether baselines are useful.

import numpy as np

# Make results repeatable for random baselines.
rng = np.random.default_rng(7)

# Create small regression targets in kilowatt hours.
energy_kwh = np.array([92, 88, 95, 110, 87, 91, 160, 89])

# Validate the regression target size.
assert energy_kwh.size >= 4

# Build constant regression baseline predictions.
mean_pred = np.full(energy_kwh.shape, energy_kwh.mean())
median_pred = np.full(energy_kwh.shape, np.median(energy_kwh))

# Define mean squared error manually.
def mse(actual_values, predicted_values):
    errors = actual_values - predicted_values
    squared_errors = errors ** 2
    return squared_errors.mean()

# Create small classification labels for inspection.
actual_parts = np.array(["pass", "pass", "fail", "pass", "pass", "fail"])

# Validate the classification target size.
assert actual_parts.size >= 4

# Build majority and stratified classification baselines.
labels, counts = np.unique(actual_parts, return_counts=True)
majority_label = labels[np.argmax(counts)]
majority_pred = np.full(actual_parts.shape, majority_label)

# Sample labels using observed class proportions.
probabilities = counts / counts.sum()
stratified_pred = rng.choice(labels, size=actual_parts.size, p=probabilities)

# Define accuracy manually for classification.
def accuracy(actual_values, predicted_values):
    correct_flags = actual_values == predicted_values
    return correct_flags.mean()

# Print compact baseline comparison results.
print("Regression mean baseline MSE:", round(mse(energy_kwh, mean_pred), 2))
print("Regression median baseline MSE:", round(mse(energy_kwh, median_pred), 2))
print("Classification majority baseline accuracy:", round(accuracy(actual_parts, majority_pred), 2))
print("Classification stratified baseline accuracy:", round(accuracy(actual_parts, stratified_pred), 2))
print("Engineering note: rare failures need more than accuracy.")

```

## **2. Manual Metric Calculations**

### **2.1. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1783661244" width="250">



>* Average squared differences between predictions and truth
>* Large errors receive stronger engineering penalties

>* Square errors, then average across observations
>* Large misses reveal operational disruption risks

>* MSE uses less intuitive squared units
>* Judge improvements by engineering consequences



```
#@title Python Code - Mean Squared Error

# Manual metrics support engineering model evaluation.
# Mean squared error penalizes large misses.
# This example uses machine cycle minutes.

# Store actual cycle times in minutes.
actual_minutes = [10.0, 12.0, 9.0, 15.0, 11.0]

# Store predicted cycle times in minutes.
predicted_minutes = [9.5, 13.0, 8.0, 18.0, 10.5]

# Validate matching list lengths before calculation.
if len(actual_minutes) != len(predicted_minutes):
    raise ValueError("Lists must have matching lengths")

# Validate at least one observation exists.
if len(actual_minutes) == 0:
    raise ValueError("At least one observation is required")

# Compute prediction errors manually.
errors = []
for actual, predicted in zip(actual_minutes, predicted_minutes):
    errors.append(predicted - actual)

# Square each error manually.
squared_errors = []
for error in errors:
    squared_errors.append(error ** 2)

# Average squared errors manually.
count = len(squared_errors)
mean_squared_error = sum(squared_errors) / count

# Convert MSE into original units for intuition.
root_mean_squared_error = mean_squared_error ** 0.5

# Print compact teaching results.
print("Actual minutes:", actual_minutes)
print("Predicted minutes:", predicted_minutes)
print("Errors:", errors)
print("Squared errors:", squared_errors)
print("Manual MSE:", round(mean_squared_error, 3), "squared minutes")
print("RMSE for intuition:", round(root_mean_squared_error, 3), "minutes")
print("Large misses raise MSE and may disrupt production schedules")

```

### **2.2. Classification Metrics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1783661246" width="250">



>* Compare each prediction with its true label
>* Accuracy counts correct predictions among all cases

>* Confusion matrices sort predictions by outcome
>* Define positive class before counting errors

>* Imbalanced classes can hide poor performance
>* Error types affect engineering risk and cost



```
#@title Python Code - Classification Metrics

# Manual metrics reveal classification decision quality.
# Engineering errors can have unequal consequences.
# This example counts outcomes by hand.

# Define true labels for inspected parts.
actual_labels = [1, 0, 1, 0, 1, 0, 0, 1]

# Define predicted labels from inspection system.
predicted_labels = [1, 0, 0, 0, 1, 1, 0, 1]

# Validate both lists before metric calculations.
assert len(actual_labels) == len(predicted_labels), "Label lists must match."

# Count confusion matrix outcomes manually.
true_positive = 0
true_negative = 0
false_positive = 0

false_negative = 0

# Compare every actual and predicted label.
for actual, predicted in zip(actual_labels, predicted_labels):
    if actual == 1 and predicted == 1:
        true_positive = true_positive + 1

    elif actual == 0 and predicted == 0:
        true_negative = true_negative + 1

    elif actual == 0 and predicted == 1:
        false_positive = false_positive + 1

    elif actual == 1 and predicted == 0:
        false_negative = false_negative + 1

# Compute totals and safe denominators.
total_cases = len(actual_labels)
correct_cases = true_positive + true_negative
predicted_positive = true_positive + false_positive

actual_positive = true_positive + false_negative

# Calculate common classification metrics manually.
accuracy = correct_cases / total_cases
precision = true_positive / predicted_positive if predicted_positive else 0
recall = true_positive / actual_positive if actual_positive else 0

# Calculate F1 using precision and recall.
f1_denominator = precision + recall
f1_score = 2 * precision * recall / f1_denominator if f1_denominator else 0

# Print compact results for engineering interpretation.
print("Confusion matrix counts: TP=", true_positive, "TN=", true_negative)
print("Confusion matrix counts: FP=", false_positive, "FN=", false_negative)
print("Accuracy:", round(accuracy, 3))
print("Precision:", round(precision, 3))

print("Recall:", round(recall, 3))
print("F1 score:", round(f1_score, 3))
print("False positives may waste inspection labor.")
print("False negatives may allow defective parts through.")

```

### **2.3. Precision Recall F Score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1783661248" width="250">



>* Use confusion matrix counts for precision
>* High precision reduces costly false alarms

>* Recall measures found actual positives
>* Thresholds trade recall against precision

>* F1 balances precision and recall tradeoffs
>* Interpret metrics using positive class and risk



```
#@title Python Code - Precision Recall F Score

# Manual metrics reveal engineering decision tradeoffs.
# Positive class means machines needing maintenance.
# Counts drive precision recall and F1.

# Store true equipment states as labels.
y_true = [1, 0, 1, 1, 0, 0, 1, 0]

# Store model maintenance alert predictions.
y_pred = [1, 0, 0, 1, 1, 0, 1, 0]

# Validate equal list lengths before counting.
assert len(y_true) == len(y_pred), "Label lists must match."

# Count confusion matrix entries manually.
tp = sum(1 for t, p in zip(y_true, y_pred) if t == 1 and p == 1)
fp = sum(1 for t, p in zip(y_true, y_pred) if t == 0 and p == 1)
fn = sum(1 for t, p in zip(y_true, y_pred) if t == 1 and p == 0)
tn = sum(1 for t, p in zip(y_true, y_pred) if t == 0 and p == 0)

# Define safe division for zero denominators.
def safe_divide(numerator, denominator):
    return numerator / denominator if denominator else 0.0

# Calculate precision recall and F1.
precision = safe_divide(tp, tp + fp)
recall = safe_divide(tp, tp + fn)
f1_score = safe_divide(2 * precision * recall, precision + recall)

# Print compact engineering interpretation results.
print(f"Confusion counts: TP={tp}, FP={fp}, FN={fn}, TN={tn}")
print(f"Precision: {precision:.2f} reliability of maintenance alerts")
print(f"Recall: {recall:.2f} fraction of real failures caught")
print(f"F1 score: {f1_score:.2f} balance of alert quality")
print("False positives waste inspections; false negatives risk downtime.")

```

## **3. Engineering Risk Meaning**

### **3.1. Failure Risk**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1783661250" width="250">



>* Metrics reveal possible engineering failure consequences
>* Judge errors by safety, cost, and trust

>* Error types carry different engineering risks
>* Average metrics can hide severe failures

>* Judge errors by decision impact
>* Check cascades, baselines, and safeguards



```
#@title Python Code - Failure Risk

# Metrics reveal engineering failure risk.
# Missed failures can be costly.
# Baselines help judge decision quality.

import numpy as np

# Store true pump failure labels.
y_true = np.array([1, 0, 1, 0, 1, 0, 0, 1])

# Compare two simple inspection policies.
y_safe = np.array([1, 1, 1, 0, 1, 0, 1, 1])
y_risky = np.array([0, 0, 1, 0, 0, 0, 0, 1])

# Validate equal label lengths.
assert y_true.shape == y_safe.shape == y_risky.shape

# Define manual classification metrics.
def metrics(y, p):
    tp = int(np.sum((y == 1) & (p == 1)))
    tn = int(np.sum((y == 0) & (p == 0)))

    fp = int(np.sum((y == 0) & (p == 1)))
    fn = int(np.sum((y == 1) & (p == 0)))

    accuracy = (tp + tn) / len(y)
    precision = tp / max(tp + fp, 1)

    recall = tp / max(tp + fn, 1)
    f1 = 2 * precision * recall / max(precision + recall, 1e-9)

    return tp, tn, fp, fn, accuracy, precision, recall, f1

# Compute metrics for both policies.
safe = metrics(y_true, y_safe)
risky = metrics(y_true, y_risky)

# Assign simple engineering costs.
false_negative_cost = 100000
false_positive_cost = 8000

# Convert mistakes into expected cost.
safe_cost = safe[3] * false_negative_cost + safe[2] * false_positive_cost
risky_cost = risky[3] * false_negative_cost + risky[2] * false_positive_cost

# Print compact decision-focused results.
print("Policy        Acc  Prec Recall F1   FN FP Cost")
print(f"Conservative {safe[4]:.2f} {safe[5]:.2f} {safe[6]:.2f} {safe[7]:.2f}  {safe[3]}  {safe[2]} ${safe_cost:,}")
print(f"Risky        {risky[4]:.2f} {risky[5]:.2f} {risky[6]:.2f} {risky[7]:.2f}  {risky[3]}  {risky[2]} ${risky_cost:,}")
print("Interpretation: missed failures dominate cost and safety risk.")

```

### **3.2. Cost Sensitive Metrics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1783661252" width="250">



>* Different errors carry different engineering consequences
>* Judge models by cost, safety, and reliability

>* Best model depends on error costs
>* Balance precision and recall by application

>* Make model decision assumptions explicit
>* Map confusion matrix outcomes to real risks



```
#@title Python Code - Cost Sensitive Metrics

# Cost sensitive metrics connect errors to engineering consequences.
# This example compares two turbine warning models.
# Manual calculations keep the evaluation transparent.

# No extra installations are required.

# Store true turbine fault labels.
y_true = [1, 0, 1, 0, 1, 0, 0, 1]

# Store predictions from two warning models.
model_a = [1, 1, 0, 0, 1, 0, 1, 0]
model_b = [1, 1, 1, 0, 1, 1, 0, 1]

# Define costs for operational consequences.
false_alarm_cost = 800
missed_fault_cost = 12000

# Count confusion matrix cells manually.
def confusion_counts(actual, predicted):
    assert len(actual) == len(predicted)
    tp = sum(a == 1 and p == 1 for a, p in zip(actual, predicted))
    tn = sum(a == 0 and p == 0 for a, p in zip(actual, predicted))

    fp = sum(a == 0 and p == 1 for a, p in zip(actual, predicted))
    fn = sum(a == 1 and p == 0 for a, p in zip(actual, predicted))
    return tp, tn, fp, fn

# Compute common metrics and total cost.
def evaluate_model(name, actual, predicted):
    tp, tn, fp, fn = confusion_counts(actual, predicted)
    total = len(actual)
    accuracy = (tp + tn) / total

    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

    cost = fp * false_alarm_cost + fn * missed_fault_cost
    print(f"{name}: acc={accuracy:.2f}, precision={precision:.2f}, recall={recall:.2f}, F1={f1:.2f}")
    print(f"{name}: FP={fp}, FN={fn}, estimated cost=${cost:,}")
    return cost

# Evaluate both models using the same cost assumptions.
cost_a = evaluate_model("Model A", y_true, model_a)
cost_b = evaluate_model("Model B", y_true, model_b)

# Choose the model with lower engineering cost.
better = "Model A" if cost_a < cost_b else "Model B"
print(f"Cost-sensitive choice: {better}")
print("Missed faults dominate because downtime risk is expensive.")

```

### **3.3. Decision Threshold Tradeoffs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1783661254" width="250">



>* Thresholds turn model scores into actions
>* Lower thresholds trade missed failures for false alarms

>* Lower thresholds reduce dangerous missed failures
>* Higher thresholds reduce costly false alarms

>* Evaluate thresholds within real engineering workflows
>* Choose metrics that support responsible decisions



```
#@title Python Code - Decision Threshold Tradeoffs

# Thresholds turn scores into engineering actions.
# Lower thresholds catch more risky failures.
# Higher thresholds reduce costly false alarms.

import numpy as np
import matplotlib.pyplot as plt

# Store true pump failure outcomes.
y_true = np.array([1, 0, 1, 0, 1, 0, 0, 1])

# Store model failure risk scores.
scores = np.array([0.82, 0.62, 0.55, 0.48, 0.41, 0.35, 0.22, 0.18])

# Confirm arrays match before evaluation.
assert y_true.shape == scores.shape

# Define thresholds engineers might choose.
thresholds = np.array([0.30, 0.50, 0.70])

# Create a compact metric function.
def metrics_at_threshold(threshold):
    y_pred = (scores >= threshold).astype(int)
    tp = int(np.sum((y_true == 1) & (y_pred == 1)))
    fp = int(np.sum((y_true == 0) & (y_pred == 1)))

    fn = int(np.sum((y_true == 1) & (y_pred == 0)))
    tn = int(np.sum((y_true == 0) & (y_pred == 0)))
    precision = tp / max(tp + fp, 1)
    recall = tp / max(tp + fn, 1)

    return precision, recall, fp, fn

# Evaluate each threshold choice.
rows = [metrics_at_threshold(t) for t in thresholds]

# Print a small decision table.
print("Threshold | Precision | Recall | False alarms | Missed failures")

# Show engineering tradeoffs clearly.
for threshold, row in zip(thresholds, rows):
    precision, recall, fp, fn = row
    print(f"{threshold:8.2f} | {precision:9.2f} | {recall:6.2f} | {fp:12d} | {fn:15d}")

# Plot precision and recall tradeoff.
precision_values = [row[0] for row in rows]
recall_values = [row[1] for row in rows]

# Create one simple comparison plot.
plt.figure(figsize=(6, 4))
plt.plot(thresholds, precision_values, marker="o", label="Precision")
plt.plot(thresholds, recall_values, marker="s", label="Recall")

# Label the engineering decision axes.
plt.xlabel("Inspection threshold for pump failure risk")
plt.ylabel("Metric value")
plt.title("Threshold choice changes engineering risk")

# Finish the single visible plot.
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

```
