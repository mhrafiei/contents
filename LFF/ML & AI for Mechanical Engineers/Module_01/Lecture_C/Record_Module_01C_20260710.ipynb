### **1.1. Array Creation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1783658031" width="250">



>* Represent engineering data as structured NumPy arrays
>* Choose dimensions to match data organization

>* Create arrays from measured or generated data
>* Use filled arrays as calculation starting points

>* Choose shape, dimensions, and data type carefully
>* Match array structure to the engineering problem



```
#@title Python Code - Array Creation

# Create arrays for engineering measurements.
# Inspect shapes, dimensions, and types.
# Visualize an image-like temperature grid.

# Import NumPy and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create one-dimensional measured beam temperatures.
beam_cm = np.array([0, 25, 50, 75, 100])
temp_c = np.array([22.0, 24.5, 29.0, 33.5, 37.0])

# Create regularly spaced pipe positions.
pipe_ft = np.linspace(0, 12, 7)
pressure_psi = np.full(pipe_ft.shape, 55.0)

# Create an image-like thermal grid.
thermal_grid = np.zeros((4, 5), dtype=float)
thermal_grid[1:3, 2:4] = 85.0

# Validate expected array shapes.
assert beam_cm.shape == temp_c.shape
assert thermal_grid.shape == (4, 5)

# Print compact array summaries only.
print("Beam positions shape:", beam_cm.shape)
print("Temperature dtype:", temp_c.dtype)
print("Pipe pressure values:", pressure_psi[:3], "...")
print("Thermal grid shape:", thermal_grid.shape)
print("Thermal grid maximum:", thermal_grid.max())

# Plot the small image-like grid.
plt.figure(figsize=(5, 3))
plt.imshow(thermal_grid, cmap="hot", origin="lower")
plt.colorbar(label="Temperature (C)")
plt.title("Created NumPy Thermal Grid")
plt.xlabel("Column index")
plt.ylabel("Row index")
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Array Creation

# Create arrays for engineering calculations.
# Compare one dimensional and grid data.
# Keep outputs small and readable.

# numpy is preinstalled in Google Colab.
import numpy as np
import matplotlib.pyplot as plt

# Create measured strain readings in microstrain.
strain_micro = np.array([120, 135, 128, 142, 150])

# Create beam positions along one meter.
positions_m = np.linspace(0.0, 1.0, 5)

# Create placeholder deflection values in millimeters.
deflection_mm = np.zeros_like(positions_m)

# Create a small thermal image grid.
temperature_c = np.array([
    [22.0, 23.5, 25.0],
    [24.0, 31.0, 26.5],
    [23.0, 24.5, 25.5]
])

# Validate expected array shapes before plotting.
assert strain_micro.shape == positions_m.shape
assert temperature_c.shape == (3, 3)

# Print compact summaries, not full large arrays.
print("Strain array shape:", strain_micro.shape)
print("Position array shape:", positions_m.shape)
print("Deflection placeholder shape:", deflection_mm.shape)
print("Thermal grid shape:", temperature_c.shape)
print("Maximum temperature Celsius:", temperature_c.max())

# Plot the image-like temperature grid.
plt.figure(figsize=(4, 3))
plt.imshow(temperature_c, cmap="hot", origin="lower")
plt.colorbar(label="Temperature (C)")
plt.title("Small Thermal Grid Array")
plt.xlabel("Column index")
plt.ylabel("Row index")
plt.tight_layout()
plt.show()

```

### **1.3. Array Properties**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1783658036" width="250">



>* Shape shows array size across dimensions
>* Unexpected shapes can cause misleading calculations

>* Data type controls values, memory, and precision
>* Check types to avoid calculation mistakes

>* Check size and memory for large arrays
>* Confirm dimensions match real engineering meanings



```
#@title Python Code - Array Properties

# Explore NumPy array properties for engineering data.
# Check shape, dimensions, type, and memory.
# Use small arrays to keep output readable.

import numpy as np

# Create a one dimensional sensor vector.
temperatures_c = np.array([21.5, 22.1, 22.8, 23.0])

# Create a two dimensional pressure grid.
pressure_kpa = np.array([[101.2, 101.5, 101.7], [100.9, 101.1, 101.4]])

# Create an image like grayscale grid.
pixel_grid = np.array([[0, 80, 160], [40, 120, 255]], dtype=np.uint8)

# Store arrays with engineering labels.
arrays = [("temperature vector", temperatures_c), ("pressure grid", pressure_kpa), ("pixel grid", pixel_grid)]

# Print compact property summaries only.
for label, array in arrays:
    print(f"{label}: shape={array.shape}, ndim={array.ndim}, dtype={array.dtype}")

# Compare total values and memory usage.
for label, array in arrays:
    print(f"{label}: size={array.size}, memory={array.nbytes} bytes")

# Validate expected pressure grid shape.
expected_shape = (2, 3)

# Report whether the grid matches expectations.
shape_ok = pressure_kpa.shape == expected_shape

# Show one safe engineering calculation.
mean_pressure = pressure_kpa.mean()

# Print validation and summary result.
print(f"pressure shape expected {expected_shape}: {shape_ok}")
print(f"mean pressure: {mean_pressure:.2f} kPa")

```

## **2. pandas Tables**

### **2.1. DataFrame Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1783658047" width="250">



>* DataFrames organize engineering data into tables
>* Rows stay aligned with measured variables

>* Column labels name each measured variable
>* Row indexes identify records and preserve context

>* Inspect structure before analyzing DataFrames
>* Check types, gaps, units, and errors



```
#@title Python Code - DataFrame Basics

# DataFrames organize engineering measurements into labeled tables.
# Inspection reveals structure before analysis begins.
# Small examples keep outputs readable and focused.

import pandas as pd
import numpy as np

# Create a tiny mechanical test dataset.
data = {
    "specimen": ["A1", "A2", "B1", "B2", "C1"],
    "material": ["steel", "steel", "aluminum", "aluminum", "steel"],

    "load_kN": [12.5, 15.2, 8.7, 9.1, np.nan],
    "deflection_mm": [1.8, 2.1, 2.9, 3.2, 2.0],
    "temperature_C": [22, 24, 23, 25, 24],
}

# Build the DataFrame from labeled columns.
df = pd.DataFrame(data)

# Use specimen identifiers as row labels.
df = df.set_index("specimen")

# Inspect size, labels, and data types.
print("Shape:", df.shape)
print("Columns:", list(df.columns))
print("Index labels:", list(df.index))
print("Data types:", df.dtypes.to_dict())

# Preview only the first few records.
print("First rows:")
print(df.head(3))

# Check missing values by column.
print("Missing values:", df.isna().sum().to_dict())

# Summarize numerical engineering measurements.
summary = df[["load_kN", "deflection_mm"]].describe()
print("Mean load kN:", round(summary.loc["mean", "load_kN"], 2))
print("Max deflection mm:", round(summary.loc["max", "deflection_mm"], 2))

# Filter rows using a physical condition.
high_deflection = df[df["deflection_mm"] > 2.5]
print("High deflection specimens:", list(high_deflection.index))

```

### **2.2. Filtering DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1783658049" width="250">



>* Select relevant rows, columns, or records
>* Focus analysis on meaningful engineering subsets

>* Filter rows using column-based conditions
>* Combine conditions for engineering context

>* Choose filters with engineering context.
>* Reveal trends while avoiding hidden mistakes.



```
#@title Python Code - Filtering DataFrames

# Filtering tables supports focused engineering analysis.
# Small datasets keep outputs readable.
# Conditions select rows with meaning.

# Import pandas for tabular data handling.
import pandas as pd

# Create a small mechanical inspection dataset.
data = {
    "part_id": ["A01", "A02", "A03", "A04", "A05", "A06"],
    "material": ["Aluminum", "Steel", "Aluminum", "Steel", "Titanium", "Aluminum"],

    "diameter_in": [1.002, 0.997, 1.011, 1.000, 0.994, 1.006],
    "load_kN": [12.5, 18.2, 13.1, 21.0, 16.8, 14.4],
    "temperature_C": [24, 26, 82, 88, 25, 79],
}

# Convert the dictionary into a DataFrame.
df = pd.DataFrame(data)

# Define engineering limits for filtering.
lower_limit = 0.998
upper_limit = 1.008
hot_limit = 75

# Filter rows outside the diameter tolerance.
out_of_tolerance = df[
    (df["diameter_in"] < lower_limit) | (df["diameter_in"] > upper_limit)
]

# Filter aluminum parts tested at high temperature.
hot_aluminum = df[
    (df["material"] == "Aluminum") & (df["temperature_C"] > hot_limit)
]

# Summarize the filtered engineering subsets.
print("Total inspected parts:", len(df))
print("Out-of-tolerance parts:", len(out_of_tolerance))
print("Hot aluminum tests:", len(hot_aluminum))
print("Out-of-tolerance IDs:", out_of_tolerance["part_id"].tolist())
print("Hot aluminum IDs:", hot_aluminum["part_id"].tolist())

# Show selected columns from one filtered table.
print(hot_aluminum[["part_id", "diameter_in", "temperature_C"]].to_string(index=False))

```

### **2.3. Summarizing DataFrames**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1783658054" width="250">



>* Condense raw measurements into useful statistics
>* Spot data issues before plotting or modeling

>* Group DataFrames by meaningful engineering categories
>* Reveal hidden patterns across conditions

>* Interpret summaries with context and caution
>* Iteratively check groups, outliers, and missing data



```
#@title Python Code - Summarizing DataFrames

# Summaries turn tables into engineering insight.
# Grouped summaries compare meaningful operating conditions.
# Small tables keep beginner examples readable.

# Import pandas for tabular engineering data.
import pandas as pd

# Create a small mechanical inspection dataset.
data = {
    "part_id": [101, 102, 103, 104, 105, 106],

    "line": ["A", "A", "B", "B", "C", "C"],
    "diameter_mm": [24.98, 25.04, 25.10, 24.91, 25.02, 25.30],
    "load_kN": [12.1, 12.4, 11.8, 13.2, 12.0, 15.6],

    "passed": [True, True, True, False, True, False],
}

# Convert the dictionary into a DataFrame.
df = pd.DataFrame(data)

# Check the table size before summarizing.
assert df.shape == (6, 5)

# Summarize important numeric measurement columns.
summary = df[["diameter_mm", "load_kN"]].agg(
    ["count", "mean", "min", "max", "std"]
)

# Summarize measurements by production line.
line_summary = df.groupby("line").agg(
    mean_diameter_mm=("diameter_mm", "mean"),
    max_load_kN=("load_kN", "max"),
    pass_rate=("passed", "mean"),
)

# Print compact summaries for quick inspection.
print("Overall numeric summary:")
print(summary.round(2))

# Print grouped summaries for engineering comparison.
print("\nGrouped summary by production line:")
print(line_summary.round(2))

# Identify possible outliers using a simple threshold.
outliers = df[df["diameter_mm"] > 25.20]

# Print only the key outlier columns.
print("\nDiameter values above 25.20 mm:")
print(outliers[["part_id", "line", "diameter_mm"]].to_string(index=False))

```

## **3. Engineering Data Plots**

### **3.1. Trend Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1783658042" width="250">



>* Plots turn measurements into clear patterns
>* Trends reveal behavior, faults, and causes

>* Choose axes that match physical meaning
>* Show changes more clearly than tables

>* Spot outliers, noise, and system changes
>* Use trends to guide engineering decisions



```
#@title Python Code - Trend Visualization

# Trend plots reveal engineering behavior.
# Small datasets keep examples understandable.
# Labels connect plots to physics.

# Import numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create repeatable synthetic test data.
np.random.seed(7)
time_min = np.arange(0, 31, 2)
base_temp_c = 24 + 0.9 * time_min

# Add small sensor noise.
noise_c = np.random.normal(0, 0.8, time_min.size)
temp_c = base_temp_c + noise_c
power_kw = 4.8 + 0.04 * time_min

# Store measurements in a table.
data = pd.DataFrame({
    "time_min": time_min,
    "motor_temp_c": temp_c,
    "power_kw": power_kw,
})

# Check the table size safely.
rows, columns = data.shape
assert rows > 3 and columns == 3

# Summarize the trend with few lines.
start_temp = data["motor_temp_c"].iloc[0]
end_temp = data["motor_temp_c"].iloc[-1]
temp_change = end_temp - start_temp

# Print compact engineering interpretation.
print(f"Measurements: {rows} time samples")
print(f"Temperature rise: {temp_change:.1f} deg C")
print("Trend: motor temperature increases during the test")

# Build one clear trend plot.
plt.figure(figsize=(8, 4.5))
plt.plot(data["time_min"], data["motor_temp_c"], marker="o")
plt.xlabel("Time during test (min)")
plt.ylabel("Motor temperature (deg C)")

# Add title and grid for readability.
plt.title("Trend Visualization: Motor Temperature Test")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **3.2. Trend Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1783658040" width="250">



>* Line plots reveal changes in ordered measurements
>* Plots help diagnose hidden engineering patterns

>* Label axes clearly with meaningful units
>* Choose honest scales that show real changes

>* Compare conditions, components, and test results
>* Spot anomalies and guide engineering decisions



```
#@title Python Code - Trend Visualization

# Trend plots reveal engineering behavior.
# Small datasets keep examples understandable.
# Clear labels support engineering decisions.

# Import numerical and plotting libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create ordered time measurements.
time_hours = np.arange(0, 9, 1)
temperature_c = np.array([22, 31, 39, 46, 52, 57, 61, 64, 66])
vibration_mm_s = np.array([1.2, 1.3, 1.4, 1.6, 1.7, 2.4, 1.9, 2.0, 2.1])

# Store measurements in a table.
data = pd.DataFrame({
    "time_hours": time_hours,
    "temperature_c": temperature_c,
    "vibration_mm_s": vibration_mm_s,
})

# Check the table size safely.
row_count = len(data)
column_count = len(data.columns)
assert row_count == 9 and column_count == 3

# Summarize key trend information.
temp_change = data["temperature_c"].iloc[-1] - data["temperature_c"].iloc[0]
peak_vibration = data["vibration_mm_s"].max()
peak_time = data.loc[data["vibration_mm_s"].idxmax(), "time_hours"]

# Print only concise engineering observations.
print(f"Rows: {row_count}, columns: {column_count}")
print(f"Temperature rise: {temp_change} °C over {time_hours[-1]} hours")
print(f"Peak vibration: {peak_vibration} mm/s at hour {peak_time}")

# Create one trend visualization figure.
fig, ax1 = plt.subplots(figsize=(8, 4.5))
ax1.plot(data["time_hours"], data["temperature_c"], marker="o", color="tab:red")
ax1.set_xlabel("Operating time (hours)")
ax1.set_ylabel("Bearing temperature (°C)", color="tab:red")

# Add a second response axis.
ax2 = ax1.twinx()
ax2.plot(data["time_hours"], data["vibration_mm_s"], marker="s", color="tab:blue")
ax2.set_ylabel("Vibration velocity (mm/s)", color="tab:blue")

# Highlight the vibration anomaly.
ax2.scatter([peak_time], [peak_vibration], color="black", zorder=5)
ax2.annotate("possible anomaly", xy=(peak_time, peak_vibration), xytext=(5.2, 2.55), arrowprops={"arrowstyle": "->"})

# Finish the plot formatting.
ax1.set_title("Machine Trend Visualization During Operation")
ax1.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()

```

### **3.3. Relationship Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1783658045" width="250">



>* Scatter plots show paired engineering measurements
>* Point patterns reveal relationships or noise

>* Spot clusters and operating patterns
>* Investigate outliers for possible engineering issues

>* Choose axes, scales, and categories carefully
>* Use clear labels to communicate evidence



```
#@title Python Code - Relationship Plots

# Relationship plots reveal engineering variable patterns.
# Scatter plots compare paired measurement values.
# Colors can separate operating conditions clearly.

# Import numerical, table, and plotting tools.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create deterministic synthetic pump test data.
rng = np.random.default_rng(seed=42)
flow_gpm = np.array([20, 28, 35, 42, 50, 58, 65, 72, 80, 88, 95, 102])
valve_setting = np.array(["low", "low", "low", "low", "medium", "medium", "medium", "medium", "high", "high", "high", "high"])

# Model pressure drop with realistic noise.
base_pressure = 0.018 * flow_gpm ** 2
setting_offset = np.where(valve_setting == "high", 18, 0)
setting_offset = np.where(valve_setting == "medium", 8, setting_offset)
noise = rng.normal(loc=0, scale=3, size=flow_gpm.size)

# Store paired measurements in a DataFrame.
pressure_psi = base_pressure + setting_offset + noise
pump_data = pd.DataFrame({"flow_gpm": flow_gpm, "pressure_psi": pressure_psi, "valve_setting": valve_setting})
assert pump_data.shape == (12, 3)

# Print a compact summary for learners.
print("Rows and columns:", pump_data.shape)
print("Flow range, gpm:", pump_data["flow_gpm"].min(), "to", pump_data["flow_gpm"].max())
print("Pressure mean, psi:", round(pump_data["pressure_psi"].mean(), 1))

# Prepare one relationship plot.
fig, ax = plt.subplots(figsize=(7, 4.5))
colors = {"low": "tab:blue", "medium": "tab:orange", "high": "tab:green"}
markers = {"low": "o", "medium": "s", "high": "^"}

# Plot each valve setting separately.
for setting, group in pump_data.groupby("valve_setting"):
    ax.scatter(group["flow_gpm"], group["pressure_psi"], label=setting, color=colors[setting], marker=markers[setting], s=80)

# Add labels that include engineering units.
ax.set_title("Pump Relationship Plot: Flow Rate vs Pressure Drop")
ax.set_xlabel("Flow rate (gallons per minute)")
ax.set_ylabel("Pressure drop (psi)")
ax.grid(True, alpha=0.3)

# Add legend and display the figure.
ax.legend(title="Valve setting")
plt.tight_layout()
plt.show()

```
