### **1.1. Unit Conversion Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1783656947" width="250">



>* Convert measurements into analysis-ready units
>* Named functions improve accuracy and readability

>* Name conversions to show direction clearly
>* Make conversions testable, auditable, and reusable

>* Standardize instrument data before feature calculations
>* Check converted values to catch unit mistakes



```
#@title Python Code - Unit Conversion Functions

# Readable functions make conversions safer.
# Engineering data often mixes units.
# Small checks catch large mistakes.

# Define clear unit conversion functions.
def inches_to_meters(inches_value):
    return inches_value * 0.0254

# Define another common conversion function.
def psi_to_pascal(psi_value):
    return psi_value * 6894.757

# Define a rotational speed conversion.
def rpm_to_rad_per_second(rpm_value):
    return rpm_value * 2.0 * 3.141592653589793 / 60.0

# Store raw measurements in dictionaries.
raw_tests = [
    {"test": "A", "length_in": 2.0, "pressure_psi": 30.0, "speed_rpm": 1200.0},
    {"test": "B", "length_in": 3.5, "pressure_psi": 45.0, "speed_rpm": 1800.0},
]

# Check the small dataset size.
if len(raw_tests) == 0:
    raise ValueError("No test records were provided.")

# Convert each record into analysis units.
converted_tests = []
for row in raw_tests:
    converted_row = {
        "test": row["test"],
        "length_m": inches_to_meters(row["length_in"]),
        "pressure_pa": psi_to_pascal(row["pressure_psi"]),
        "speed_rad_s": rpm_to_rad_per_second(row["speed_rpm"]),
    }

    converted_tests.append(converted_row)

# Print compact debugging information.
print("records converted:", len(converted_tests))
print("first raw length inches:", raw_tests[0]["length_in"])
print("first converted length meters:", round(converted_tests[0]["length_m"], 4))
print("first converted pressure kPa:", round(converted_tests[0]["pressure_pa"] / 1000.0, 1))
print("first converted speed rad/s:", round(converted_tests[0]["speed_rad_s"], 1))

# Verify converted values are physically plausible.
lengths_are_positive = all(row["length_m"] > 0.0 for row in converted_tests)
pressures_are_positive = all(row["pressure_pa"] > 0.0 for row in converted_tests)
print("positive converted values:", lengths_are_positive and pressures_are_positive)

```

### **1.2. Feature Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1783656949" width="250">



>* Turn raw measurements into useful engineering features
>* Name calculations by their engineering meaning

>* Separate engineering logic from data handling
>* Name features to show purpose clearly

>* Check assumptions and handle messy measurements
>* Create reliable, understandable engineering features



```
#@title Python Code - Feature Functions

# Feature functions turn measurements into useful engineering quantities.
# Clear names make calculations easier to review.
# Small checks help catch data problems early.

# Import pandas for small tabular measurement data.
import pandas as pd

# Define a safe stress feature function.
def stress_psi(force_lbf, area_in2):
    if area_in2 <= 0:
        return None

    stress_value = force_lbf / area_in2
    return round(stress_value, 1)

# Define a motor power feature function.
def power_watts(torque_nm, speed_rpm):
    angular_speed = speed_rpm * 2 * 3.14159 / 60
    power_value = torque_nm * angular_speed
    return round(power_value, 1)

# Define a vibration peak feature function.
def peak_vibration_mm(samples_mm):
    if len(samples_mm) == 0:
        return None

    peak_value = max(abs(value) for value in samples_mm)
    return round(peak_value, 3)

# Store tiny test measurements as dictionaries.
measurements = [
    {"part": "shaft_A", "force_lbf": 1200, "area_in2": 0.75},
    {"part": "shaft_B", "force_lbf": 950, "area_in2": 0.50},
]

# Add stress features using a loop.
for row in measurements:
    row["stress_psi"] = stress_psi(row["force_lbf"], row["area_in2"])

# Create a compact table for inspection.
stress_table = pd.DataFrame(measurements)
print("Stress feature table:")
print(stress_table[["part", "stress_psi"]])

# Compute one motor power feature.
motor_power = power_watts(torque_nm=18, speed_rpm=1750)
print("Motor power feature, W:", motor_power)

# Summarize a short vibration time series.
vibration_samples = [0.04, -0.08, 0.03, 0.12, -0.05]
peak_feature = peak_vibration_mm(vibration_samples)

# Print checks for debugging intermediate results.
print("Vibration sample count:", len(vibration_samples))
print("Peak vibration feature, mm:", peak_feature)
print("Stress table shape:", stress_table.shape)

```

### **1.3. Feature Scaling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1783656955" width="250">



>* Scale measurements for fair comparison
>* Write functions with clear inputs and outputs

>* Prevent large-range features from dominating calculations
>* Scale against limits or normal operating values

>* Show scaling assumptions and reference values
>* Handle unusual readings for reliable features



```
#@title Python Code - Feature Scaling

# Readable functions make scaling assumptions visible.
# Mechanical measurements often use different units.
# Scaled features support fairer comparisons.

# No extra installations are required.

# Store small sensor records as dictionaries.
records = [
    {"rpm": 900, "temp_c": 62, "vibration_in_s": 0.08},
    {"rpm": 1500, "temp_c": 78, "vibration_in_s": 0.14},

    {"rpm": 2100, "temp_c": 95, "vibration_in_s": 0.22},
]

# Define engineering reference values clearly.
rated_speed_rpm = 2400
normal_temp_c = 70
max_vibration_in_s = 0.30

# Scale a value by a rated maximum.
def fraction_of_rating(value, rating):
    if rating <= 0:
        raise ValueError("rating must be positive")

    return value / rating

# Scale temperature as deviation from normal.
def temperature_deviation(value_c, reference_c):
    return value_c - reference_c

# Build scaled features for each record.
scaled_rows = []
for row in records:
    scaled_row = {
        "speed_fraction": fraction_of_rating(row["rpm"], rated_speed_rpm),
        "temp_deviation_c": temperature_deviation(row["temp_c"], normal_temp_c),

        "vibration_fraction": fraction_of_rating(row["vibration_in_s"], max_vibration_in_s),
    }
    scaled_rows.append(scaled_row)

# Check the expected number of transformed rows.
assert len(scaled_rows) == len(records)

# Print compact debugging information.
print("Raw rows:", len(records))
print("Scaled rows:", len(scaled_rows))
print("First scaled row:", scaled_rows[0])
print("Last speed fraction:", round(scaled_rows[-1]["speed_fraction"], 3))
print("Last vibration fraction:", round(scaled_rows[-1]["vibration_fraction"], 3))

```

## **2. Core Containers**

### **2.1. Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1783656937" width="250">



>* Store ordered engineering measurements together
>* Loop, filter, and combine list data

>* Store raw and processed measurements in lists
>* Check list alignment before analyzing data

>* Use loops to process list values consistently
>* Build cleaned datasets and stronger data intuition



```
#@title Python Code - Lists

# Lists keep ordered engineering measurements together.
# Loops transform readings consistently and safely.
# Checks reveal alignment and data issues.

# Store time and temperature readings.
time_seconds = [0, 1, 2, 3, 4, 5]
temp_celsius = [22.0, 22.4, 23.1, 24.0, 25.2, 26.5]

# Check related lists before pairing values.
same_length = len(time_seconds) == len(temp_celsius)
print("Aligned time and temperature lists:", same_length)

# Convert Celsius readings into Fahrenheit values.
temp_fahrenheit = []
for value_celsius in temp_celsius:
    value_fahrenheit = value_celsius * 9 / 5 + 32
    temp_fahrenheit.append(round(value_fahrenheit, 1))

# Inspect only small slices of each list.
print("First three Celsius readings:", temp_celsius[:3])
print("Last three Fahrenheit readings:", temp_fahrenheit[-3:])

# Find readings above a simple safety threshold.
hot_times = []
for index, value_celsius in enumerate(temp_celsius):
    if value_celsius > 24.5:
        hot_times.append(time_seconds[index])

# Summarize the derived list clearly.
print("Times above 24.5 C:", hot_times)
print("Number of hot readings:", len(hot_times))

# Pair aligned lists into small row records.
rows = []
for index in range(len(time_seconds)):
    row = [time_seconds[index], temp_celsius[index]]
    rows.append(row)

# Show compact tabular organization.
print("First row [time_s, temp_C]:", rows[0])
print("Total rows organized:", len(rows))

```

### **2.2. Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1783656940" width="250">



>* Store measurements with meaningful labels
>* Make small engineering datasets easier to read

>* Represent table rows with labeled dictionaries
>* Keep measurements linked to useful context

>* Map sensors or tests to readings
>* Use consistent labels for easier debugging



```
#@title Python Code - Dictionaries

# Dictionaries label engineering measurements clearly.
# Small records can form readable tables.
# Loops process named fields safely.

# Create one labeled inspection record.
inspection = {
    "part_id": "BRG-014",
    "material": "steel",

    "diameter_in": 2.50,
    "load_N": 1250.0,
    "temperature_C": 82.0,
    "passed": True,
}

# Retrieve values by meaningful keys.
print("Part:", inspection["part_id"])
print("Load N:", inspection["load_N"])

# Store units beside measurement names.
units = {
    "diameter_in": "inches",
    "load_N": "newtons",
    "temperature_C": "degrees Celsius",
}

# Build a tiny table of readings.
readings = [
    {"time_s": 0, "force_N": 0, "disp_mm": 0.00},
    {"time_s": 1, "force_N": 420, "disp_mm": 0.18},

    {"time_s": 2, "force_N": 860, "disp_mm": 0.39},
    {"time_s": 3, "force_N": 1250, "disp_mm": 0.61},
]

# Check table size before processing.
row_count = len(readings)
print("Rows:", row_count)

# Confirm every row has matching keys.
expected_keys = set(readings[0].keys())
keys_match = all(set(row.keys()) == expected_keys for row in readings)
print("Keys consistent:", keys_match)

# Loop through rows to compute stiffness.
stiffness_values = []
for row in readings:
    if row["disp_mm"] > 0:
        stiffness_values.append(row["force_N"] / row["disp_mm"])

# Summarize the derived feature safely.
average_stiffness = sum(stiffness_values) / len(stiffness_values)
print("Average stiffness N/mm:", round(average_stiffness, 1))

# Organize time series by sensor name.
sensor_data = {
    "thermocouple_C": [22.0, 24.5, 27.0, 29.0],
    "accelerometer_g": [0.02, 0.05, 0.04, 0.03],
}

# Debug by checking each sensor length.
for sensor_name, values in sensor_data.items():
    print(sensor_name, "samples:", len(values))

```

### **2.3. Sets**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1783656941" width="250">



>* Sets store unique values, not ordered sequences
>* Use sets to find categories and duplicates

>* Compare labels across engineering data sources
>* Validate columns before deeper analysis

>* Sets ignore order and repeated values
>* Use sets to summarize unique items



```
#@title Python Code - Sets

# Sets summarize unique engineering labels.
# Membership checks catch missing sensors.
# Comparisons reveal dataset coverage differences.

# Create repeated sensor labels from test rows.
sensor_rows = [
    "accel_x", "temp_c", "torque_nm",
    "accel_x", "rpm", "temp_c",
]

# Convert repeated labels into unique labels.
observed_sensors = set(sensor_rows)
expected_sensors = {"accel_x", "temp_c", "torque_nm", "rpm"}

# Compare expected and observed sensor coverage.
missing_sensors = expected_sensors - observed_sensors
unexpected_sensors = observed_sensors - expected_sensors

# Store repeated operating modes from measurements.
mode_rows = ["idle", "load", "load", "coast", "idle"]
unique_modes = set(mode_rows)

# Check required phases using set membership.
required_modes = {"idle", "load", "coast"}
all_modes_present = required_modes <= unique_modes

# Print compact debugging summaries only.
print("Rows recorded:", len(sensor_rows))
print("Unique sensors:", sorted(observed_sensors))
print("Missing sensors:", sorted(missing_sensors))
print("Unexpected sensors:", sorted(unexpected_sensors))
print("Unique modes:", sorted(unique_modes))
print("All required modes present:", all_modes_present)

```

## **3. Files and Debugging**

### **3.1. Read Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1783656958" width="250">



>* Treat file reading as a verification step
>* Check source, folder, contents, and format

>* Check rows, columns, and ordering
>* Catch delimiter and data type problems

>* Inspect first and last records carefully
>* Verify data before calculating engineering features



```
#@title Python Code - Read Files

# Read files with debugging checks.
# Inspect shape before trusting calculations.
# Use tiny engineering measurement data.

import csv
from pathlib import Path

# Create a small vibration sensor file.
file_path = Path("vibration_readings.csv")
rows = [
    ["time_s", "accel_g", "note"],
    ["0", "0.02", "startup"],
    ["1", "0.05", "steady"],
    ["2", "", "missing sensor value"],
    ["3", "0.08", "steady"],
]

# Write the sample file safely.
with file_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Read the file as dictionaries.
with file_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Check basic file reading results.
print(f"Loaded file: {file_path.resolve().name}")
print(f"Rows loaded: {len(data)}")
print(f"Columns found: {reader.fieldnames}")

# Inspect first and last records.
print(f"First record: {data[0]}")
print(f"Last record: {data[-1]}")

# Convert numeric values defensively.
valid_accel = []
missing_count = 0
for row in data:
    value_text = row["accel_g"].strip()
    if value_text == "":
        missing_count += 1
    else:
        valid_accel.append(float(value_text))

# Summarize intermediate debugging checks.
print(f"Valid acceleration values: {len(valid_accel)}")
print(f"Missing acceleration values: {missing_count}")
print(f"Peak acceleration: {max(valid_accel):.3f} g")

```

### **3.2. Writing Debug Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1783656959" width="250">



>* Add checks near possible data problems
>* Verify readings, order, and plausible ranges

>* Check physical limits and data structure
>* Isolate formula, unit, or input problems

>* Add checks throughout the workflow
>* Record assumptions to protect engineering conclusions



```
#@title Python Code - Writing Debug Checks

# Debug checks protect engineering data workflows.
# Small tests reveal problems early.
# This example checks vibration measurements.

# Store tiny time-series measurements.
time_s = [0.0, 0.1, 0.2, 0.3, 0.4]
accel_g = [0.02, 0.05, 0.04, 3.80, 0.03]
labels = {"time": "seconds", "accel": "g"}

# Check expected dictionary labels.
expected_keys = {"time", "accel"}
missing_keys = expected_keys - set(labels.keys())
print("Missing labels:", sorted(missing_keys))

# Check matching list lengths.
same_length = len(time_s) == len(accel_g)
print("Lengths match:", same_length)

# Check time values increase.
time_pairs = zip(time_s, time_s[1:])
time_increases = all(later > earlier for earlier, later in time_pairs)
print("Time increases:", time_increases)

# Check acceleration range plausibility.
limit_g = 2.0
bad_points = []
for index, value in enumerate(accel_g):
    if abs(value) > limit_g:
        bad_points.append((index, value))

# Report suspicious intermediate values.
print("Suspicious points:", bad_points)

# Compute feature only after checks.
if same_length and time_increases:
    peak_g = max(abs(value) for value in accel_g)
else:
    peak_g = None

# Print final feature summary.
print("Peak acceleration g:", peak_g)
print("Debug checks completed safely.")

```

### **3.3. Check Intermediate Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1783656962" width="250">



>* Check data after each major step
>* Catch scale and value errors early

>* Check dataset structure after each change
>* Catch mismatches before engineering conclusions suffer

>* Place checks after each logical step
>* Use evidence to trace debugging problems



```
#@title Python Code - Check Intermediate Results

# Checkpoints make data processing easier to debug.
# Small engineering tables reveal mistakes quickly.
# Print only compact evidence at each step.

# Import pandas for small tabular data.
import pandas as pd

# Create tiny pump test measurements.
raw_rows = [
    {"time_s": 0, "flow_gpm": 10.0, "pressure_psi": 18.0},
    {"time_s": 1, "flow_gpm": 10.5, "pressure_psi": 18.4},
    {"time_s": 2, "flow_gpm": -1.0, "pressure_psi": 18.7},
    {"time_s": 3, "flow_gpm": 11.2, "pressure_psi": 19.1},
]

# Load rows into a dataframe.
df = pd.DataFrame(raw_rows)

# Check the loaded table shape.
print("Loaded shape:", df.shape)

# Check the first two records.
print("First rows:", df.head(2).to_dict("records"))

# Confirm time values increase.
time_is_ordered = df["time_s"].is_monotonic_increasing

# Report the time ordering check.
print("Time ordered:", time_is_ordered)

# Filter physically invalid flow values.
clean_df = df[df["flow_gpm"] >= 0].copy()

# Check how many rows remain.
print("Rows after filtering:", len(clean_df))

# Convert gallons per minute to liters per second.
clean_df["flow_L_s"] = clean_df["flow_gpm"] * 0.06309

# Convert pressure psi to kilopascals.
clean_df["pressure_kPa"] = clean_df["pressure_psi"] * 6.89476

# Check representative converted values.
print("Converted sample:", clean_df.iloc[0][["flow_L_s", "pressure_kPa"]].round(2).to_dict())

# Validate expected columns before summarizing.
needed_columns = {"flow_L_s", "pressure_kPa"}

# Stop early if a required column is missing.
assert needed_columns.issubset(clean_df.columns), "Missing derived columns"

# Calculate a compact engineering summary.
summary = clean_df[["flow_L_s", "pressure_kPa"]].mean().round(2)

# Print final checked summary.
print("Mean checked features:", summary.to_dict())

```
