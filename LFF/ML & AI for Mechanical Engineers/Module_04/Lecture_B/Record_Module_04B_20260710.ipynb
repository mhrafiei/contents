## **1. Minimizing Prediction Error**

### **1.1. Error Minimization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1783664697" width="250">



>* Prediction errors guide model improvement
>* Parameters adjust repeatedly for better accuracy

>* Combine individual errors into one loss value
>* Use loss changes to improve predictions

>* Gradient descent moves downhill toward lower error
>* Repeated corrections improve predictions without understanding meaning



```
#@title Python Code - Error Minimization

# This script shows error minimization.
# A line improves through repeated corrections.
# Loss measures overall prediction error.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create deterministic mechanical test data.
np.random.seed(7)
loads = np.array([1, 2, 3, 4, 5, 6], dtype=float)
deflection = np.array([2.1, 4.0, 6.2, 7.9, 10.1, 12.2])

# Validate matching one dimensional arrays.
assert loads.ndim == deflection.ndim == 1
assert loads.size == deflection.size

# Start with an imperfect straight line.
slope = 0.2
intercept = 0.0
learning_rate = 0.04

# Store loss values for learning curve.
loss_history = []
steps = 40
sample_count = loads.size

# Repeatedly adjust parameters to reduce error.
for step in range(steps):
    predictions = slope * loads + intercept
    errors = predictions - deflection
    loss = np.mean(errors ** 2)

    loss_history.append(loss)
    slope_gradient = 2 * np.mean(errors * loads)
    intercept_gradient = 2 * np.mean(errors)
    slope = slope - learning_rate * slope_gradient

    intercept = intercept - learning_rate * intercept_gradient

# Compute final predictions and loss.
final_predictions = slope * loads + intercept
final_loss = np.mean((final_predictions - deflection) ** 2)

# Print a compact learning summary.
print(f"Initial loss: {loss_history[0]:.3f}")
print(f"Final loss: {final_loss:.3f}")
print(f"Learned line: deflection = {slope:.2f}*load + {intercept:.2f}")

# Plot data, fitted line, and loss curve.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(loads, deflection, label="observed")
axes[0].plot(loads, final_predictions, color="red", label="learned line")

axes[0].set_xlabel("Load in kN")
axes[0].set_ylabel("Deflection in mm")
axes[0].legend()
axes[1].plot(loss_history, color="purple")

axes[1].set_xlabel("Gradient descent step")
axes[1].set_ylabel("Mean squared error")
axes[1].set_title("Error decreases over repeated adjustments")
plt.tight_layout()
plt.show()

```

### **1.2. Weight Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1783664700" width="250">



>* Prediction errors guide parameter adjustments
>* Small updates gradually improve model predictions

>* Gradients guide update direction and size
>* Learning rate controls careful adjustment

>* Repeated updates align weights with data patterns
>* Errors guide smaller corrections toward better predictions



```
#@title Python Code - Weight Updates

# Weight updates turn errors into learning.
# Tiny rent data keeps calculations visible.
# Each step nudges slope and intercept.

import numpy as np
import matplotlib.pyplot as plt

# Store apartment areas and monthly rents.
area_sqft = np.array([500, 700, 900, 1100], dtype=float)
rent_dollars = np.array([1200, 1500, 1800, 2100], dtype=float)

# Scale area for stable beginner calculations.
x_values = area_sqft / 1000.0
y_values = rent_dollars / 1000.0

# Validate matching one dimensional training arrays.
assert x_values.ndim == y_values.ndim == 1
assert x_values.size == y_values.size

# Start with deliberately imperfect parameters.
weight_value = 0.20
bias_value = 0.80
learning_rate = 0.25

# Save learning curve values for plotting.
loss_history = []
weight_history = []

# Perform a few batch gradient descent updates.
for step_index in range(12):
    predictions = weight_value * x_values + bias_value
    errors = predictions - y_values

    mean_loss = np.mean(errors ** 2)
    loss_history.append(mean_loss)
    weight_history.append(weight_value)

    weight_gradient = 2.0 * np.mean(errors * x_values)
    bias_gradient = 2.0 * np.mean(errors)

    weight_value = weight_value - learning_rate * weight_gradient
    bias_value = bias_value - learning_rate * bias_gradient

# Compare the first and final update states.
print(f"Initial loss: {loss_history[0]:.4f}")
print(f"Final loss:   {loss_history[-1]:.4f}")
print(f"Learned weight: {weight_value:.3f}")
print(f"Learned intercept: {bias_value:.3f}")
print("Each update moved parameters using current error.")

# Plot loss decrease across repeated updates.
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(loss_history) + 1), loss_history, marker="o")
plt.xlabel("Weight update number")
plt.ylabel("Mean squared error")
plt.title("Prediction Error Falls After Weight Updates")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.3. Loss Landscape**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1783664703" width="250">



>* Error forms a landscape over parameters
>* Gradient descent steps downhill using local slope

>* Bowl-shaped loss guides downhill updates
>* Prediction feedback moves parameters toward lower error

>* Step size affects learning speed and stability
>* Local slope guides each parameter update



```
#@title Python Code - Loss Landscape

# Visualize loss as a parameter landscape.
# Linear regression uses slope and intercept.
# Lower landscape height means smaller error.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny mechanical calibration dataset.
np.random.seed(7)
force_newtons = np.array([10, 20, 30, 40, 50, 60], dtype=float)

# Deflection follows an approximately linear pattern.
deflection_mm = 1.8 + 0.42 * force_newtons
noise_mm = np.array([0.2, -0.1, 0.15, -0.2, 0.1, -0.05])
deflection_mm = deflection_mm + noise_mm

# Validate matching one dimensional data shapes.
assert force_newtons.ndim == deflection_mm.ndim == 1
assert force_newtons.size == deflection_mm.size

# Build candidate slopes and intercepts.
slopes = np.linspace(0.25, 0.60, 80)
intercepts = np.linspace(-2.0, 6.0, 80)
S, B = np.meshgrid(slopes, intercepts)

# Compute mean squared error for every candidate.
predictions = S[..., None] * force_newtons + B[..., None]
errors = predictions - deflection_mm
loss = np.mean(errors ** 2, axis=2)

# Find the lowest point on the landscape.
best_index = np.unravel_index(np.argmin(loss), loss.shape)
best_slope = S[best_index]
best_intercept = B[best_index]
best_loss = loss[best_index]

# Print a compact interpretation for learners.
print(f"Best slope: {best_slope:.3f} mm/N")
print(f"Best intercept: {best_intercept:.3f} mm")
print(f"Lowest mean squared error: {best_loss:.4f}")
print("Gradient descent tries to step toward this low region.")

# Draw exactly one contour plot of the landscape.
fig, ax = plt.subplots(figsize=(7, 5))
contours = ax.contourf(S, B, loss, levels=30, cmap="viridis")
fig.colorbar(contours, ax=ax, label="Mean squared error")

# Mark the best parameter setting found.
ax.scatter(best_slope, best_intercept, color="red", s=70)
ax.set_xlabel("Slope, millimeters per newton")
ax.set_ylabel("Intercept, millimeters")
ax.set_title("Loss Landscape for Linear Regression")

# Show the single teaching plot.
plt.show()

```

## **2. Descent Variants**

### **2.1. Batch Gradient Descent**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1783664690" width="250">



>* Uses all training data before updating parameters
>* Adjusts weights to reduce average prediction error

>* Update all parameters using full-dataset errors
>* Stable steps, but slow on huge datasets

>* Scale features for smoother parameter updates
>* Track loss to diagnose learning behavior



```
#@title Python Code - Batch Gradient Descent

# Batch gradient descent updates all parameters together.
# This example uses scaled apartment features.
# The learning curve reveals convergence behavior.

import numpy as np
import matplotlib.pyplot as plt

# Create deterministic apartment data in metric units.
rng = np.random.default_rng(seed=7)

# Build small feature columns for rent prediction.
area_m2 = rng.uniform(35.0, 120.0, 80)
bedrooms = rng.integers(1, 5, 80).astype(float)
distance_km = rng.uniform(0.5, 18.0, 80)

# Combine features into one design matrix.
X_raw = np.column_stack((area_m2, bedrooms, distance_km))
assert X_raw.shape == (80, 3)

# Generate rent with noise and known relationships.
noise = rng.normal(0.0, 90.0, 80)
y = 450.0 + 18.0 * area_m2 + 160.0 * bedrooms - 35.0 * distance_km + noise

# Scale features for steadier gradient descent steps.
means = X_raw.mean(axis=0)
stds = X_raw.std(axis=0)
X_scaled = (X_raw - means) / stds

# Add an intercept column of ones.
ones = np.ones((X_scaled.shape[0], 1))
X = np.column_stack((ones, X_scaled))
assert X.shape[0] == y.shape[0]

# Initialize parameters and learning settings.
theta = np.zeros(X.shape[1])
learning_rate = 0.08
epochs = 120
loss_history = []

# Run batch gradient descent using every row each update.
for epoch in range(epochs):
    predictions = X @ theta
    errors = predictions - y
    loss = np.mean(errors ** 2)
    loss_history.append(loss)

    gradient = (2.0 / X.shape[0]) * (X.T @ errors)
    theta = theta - learning_rate * gradient

# Convert scaled coefficients back to original units.
original_weights = theta[1:] / stds
original_intercept = theta[0] - np.sum(theta[1:] * means / stds)

# Print a compact summary of learning behavior.
print(f"NumPy version: {np.__version__}")
print(f"Initial MSE: {loss_history[0]:,.1f}")
print(f"Final MSE: {loss_history[-1]:,.1f}")
print(f"Intercept estimate: ${original_intercept:,.0f}")
print(f"Area weight: ${original_weights[0]:.1f} per m^2")
print(f"Bedroom weight: ${original_weights[1]:.1f} per bedroom")
print(f"Distance weight: ${original_weights[2]:.1f} per km")

# Plot the learning curve for diagnosis.
plt.figure(figsize=(7, 4))
plt.plot(loss_history, color="navy", linewidth=2)
plt.xlabel("Full-dataset update number")
plt.ylabel("Mean squared error")
plt.title("Batch Gradient Descent Learning Curve")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Stochastic Gradient Descent**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1783664695" width="250">



>* Updates parameters after each training example
>* Random shuffling enables faster large-data learning

>* Update weights after each example's error
>* Choose learning rate to balance noisy steps

>* Noisy updates can still show progress
>* Shuffle data and update incrementally



```
#@title Python Code - Stochastic Gradient Descent

# Demonstrate stochastic gradient descent from scratch.
# Use tiny delivery data for regression.
# Compare noisy updates with epoch losses.

import numpy as np
import matplotlib.pyplot as plt

# Set deterministic randomness for repeatable shuffling.
rng = np.random.default_rng(7)

# Create small delivery distance data in kilometers.
X_raw = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)

# Create delivery times with a simple linear pattern.
y = np.array([9, 12, 15, 17, 21, 23, 26, 29], dtype=float)

# Validate matching one dimensional training arrays.
assert X_raw.ndim == y.ndim == 1
assert X_raw.size == y.size

# Standardize the feature for stable gradient steps.
X = (X_raw - X_raw.mean()) / X_raw.std()

# Add an intercept column to the design matrix.
X_design = np.column_stack((np.ones(X.size), X))

# Define mean squared error for monitoring.
def mse_loss(X_matrix, y_vector, theta_vector):
    predictions = X_matrix @ theta_vector
    errors = predictions - y_vector
    return float(np.mean(errors ** 2))

# Define stochastic gradient descent from scratch.
def stochastic_gradient_descent(X_matrix, y_vector, rate, epochs):
    theta = np.zeros(X_matrix.shape[1])
    losses = []
    n_rows = X_matrix.shape[0]

    for epoch in range(epochs):
        order = rng.permutation(n_rows)
        for row_index in order:
            x_i = X_matrix[row_index]
            error = float(x_i @ theta - y_vector[row_index])
            theta = theta - rate * 2.0 * error * x_i
        losses.append(mse_loss(X_matrix, y_vector, theta))

    return theta, losses

# Train with frequent single example updates.
learning_rate = 0.04
num_epochs = 35

# Run stochastic gradient descent silently.
theta_sgd, loss_history = stochastic_gradient_descent(
    X_design, y, learning_rate, num_epochs
)

# Convert standardized slope back to original units.
slope = theta_sgd[1] / X_raw.std()
intercept = theta_sgd[0] - slope * X_raw.mean()

# Print a compact summary under fifteen lines.
print(f"NumPy version: {np.__version__}")
print(f"Learned model: time = {intercept:.2f} + {slope:.2f} * kilometers")
print(f"First epoch MSE: {loss_history[0]:.3f}")
print(f"Final epoch MSE: {loss_history[-1]:.3f}")
print("SGD loss can wiggle, but should trend downward.")

# Plot the learning curve across full passes.
plt.figure(figsize=(7, 4))
plt.plot(range(1, num_epochs + 1), loss_history, marker="o")
plt.xlabel("Epoch, one shuffled pass through all deliveries")
plt.ylabel("Mean squared error")
plt.title("Stochastic Gradient Descent Learning Curve")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **2.3. Epoch Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1783664693" width="250">



>* One epoch means one full data pass
>* Batch updates once; stochastic updates many times

>* Separate update steps from training loops
>* Shuffle each epoch to reduce ordering bias

>* Check error after each epoch
>* Use learning curves to adjust training



```
#@title Python Code - Epoch Updates

# Epochs organize repeated gradient descent updates.
# Batch and stochastic methods update differently.
# Learning curves summarize progress per epoch.

import numpy as np
import matplotlib.pyplot as plt

# Create a tiny mechanical engineering dataset.
rng = np.random.default_rng(7)

# Feature is beam length in meters.
x = np.linspace(1.0, 6.0, 24)

# Target is measured deflection in millimeters.
y = 2.5 + 1.8 * x + rng.normal(0.0, 0.35, x.size)

# Validate the one dimensional training arrays.
assert x.shape == y.shape and x.size > 2

# Scale the feature for steadier updates.
x_mean, x_std = x.mean(), x.std()
x_scaled = (x - x_mean) / x_std

# Define mean squared error for monitoring.
def mse_loss(w0, w1, x_values, y_values):
    preds = w0 + w1 * x_values
    errors = preds - y_values

    return float(np.mean(errors ** 2))

# Run batch gradient descent by epoch.
def train_batch(x_values, y_values, epochs, lr):
    w0, w1 = 0.0, 0.0
    losses = []

    for epoch in range(epochs):
        preds = w0 + w1 * x_values
        errors = preds - y_values

        grad0 = 2.0 * np.mean(errors)
        grad1 = 2.0 * np.mean(errors * x_values)

        w0 -= lr * grad0
        w1 -= lr * grad1

        losses.append(mse_loss(w0, w1, x_values, y_values))
    return w0, w1, losses

# Run stochastic gradient descent by epoch.
def train_stochastic(x_values, y_values, epochs, lr, seed):
    w0, w1 = 0.0, 0.0
    losses = []

    local_rng = np.random.default_rng(seed)
    n_samples = x_values.size

    for epoch in range(epochs):
        order = local_rng.permutation(n_samples)

        for index in order:
            pred = w0 + w1 * x_values[index]
            error = pred - y_values[index]

            w0 -= lr * 2.0 * error
            w1 -= lr * 2.0 * error * x_values[index]

        losses.append(mse_loss(w0, w1, x_values, y_values))
    return w0, w1, losses

# Train both variants for equal epoch counts.
epochs = 25
batch_w0, batch_w1, batch_losses = train_batch(x_scaled, y, epochs, 0.08)
sgd_w0, sgd_w1, sgd_losses = train_stochastic(x_scaled, y, epochs, 0.015, 11)

# Print a compact comparison of epoch behavior.
print(f"NumPy version: {np.__version__}")
print(f"Dataset size: {x.size} beams, one feature")
print(f"Batch updates per epoch: 1")
print(f"SGD updates per epoch: {x.size}")
print(f"Batch final MSE: {batch_losses[-1]:.3f}")
print(f"SGD final MSE: {sgd_losses[-1]:.3f}")

# Plot learning curves after each epoch.
plt.figure(figsize=(7, 4))
plt.plot(range(1, epochs + 1), batch_losses, marker="o", label="Batch GD")
plt.plot(range(1, epochs + 1), sgd_losses, marker="s", label="Stochastic GD")
plt.xlabel("Epoch")
plt.ylabel("Training MSE")
plt.title("Epoch Updates: Batch versus Stochastic Descent")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

## **3. Convergence Checks**

### **3.1. Learning Curve Trends**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1783664686" width="250">



>* Learning curves track loss during training
>* Healthy training trends downward despite small fluctuations

>* Error drops quickly, then improvements slow
>* Separate healthy plateaus from premature stopping

>* Judge trends by scale and noise
>* Stable decline suggests convergence; rises need diagnosis



```
#@title Python Code - Learning Curve Trends

# Learning curves reveal gradient descent behavior.
# We compare healthy, slow, and divergent training.
# Synthetic beam data keeps this example small.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create one mechanical engineering feature.
n_samples = 80
length_m = rng.uniform(1.0, 5.0, n_samples)

# Create noisy deflection measurements.
noise_mm = rng.normal(0.0, 0.8, n_samples)
deflection_mm = 2.5 + 3.2 * length_m + noise_mm

# Validate matching vector lengths.
assert length_m.shape == deflection_mm.shape
assert n_samples == length_m.size

# Standardize the feature for stable updates.
x_mean = length_m.mean()
x_std = length_m.std()
x_scaled = (length_m - x_mean) / x_std

# Add an intercept column manually.
ones = np.ones(n_samples)
X = np.column_stack((ones, x_scaled))
y = deflection_mm

# Define mean squared error loss.
def mse_loss(X, y, theta):
    predictions = X @ theta
    errors = predictions - y
    return float(np.mean(errors ** 2))

# Run batch gradient descent silently.
def batch_gd(X, y, lr, epochs):
    theta = np.zeros(X.shape[1])
    losses = []
    for epoch in range(epochs):
        errors = X @ theta - y
        gradient = (2 / len(y)) * (X.T @ errors)
        theta = theta - lr * gradient
        losses.append(mse_loss(X, y, theta))
    return np.array(losses)

# Train three learning-rate scenarios.
epochs = 60
healthy = batch_gd(X, y, lr=0.08, epochs=epochs)
slow = batch_gd(X, y, lr=0.005, epochs=epochs)
divergent = batch_gd(X, y, lr=1.05, epochs=epochs)

# Print a compact diagnostic summary.
print(f"NumPy version: {np.__version__}")
print(f"Healthy loss: {healthy[0]:.2f} to {healthy[-1]:.2f}")
print(f"Slow loss: {slow[0]:.2f} to {slow[-1]:.2f}")
print(f"Divergent loss: {divergent[0]:.2f} to {divergent[-1]:.2e}")
print("Downward then flat suggests reasonable convergence.")

# Plot the three learning curve trends.
plt.figure(figsize=(8, 5))
plt.plot(healthy, label="reasonable convergence")
plt.plot(slow, label="slow learning")
plt.plot(divergent, label="divergence")

# Use logarithmic scale to compare trends.
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Mean squared error")
plt.title("Learning Curve Trends for Gradient Descent")

# Add grid and legend for readability.
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

```

### **3.2. Divergence Detection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1783664684" width="250">



>* Rising loss signals divergence
>* Oversized updates overshoot better solutions

>* Large rates, scaling, outliers can destabilize
>* Repeated spikes or growth signal failure

>* Use curve shape and setup context
>* Stop, inspect, and adjust divergence



```
#@title Python Code - Divergence Detection

# Demonstrate divergence using linear regression.
# Compare safe and aggressive learning rates.
# Inspect learning curves for warning signs.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make the example deterministic.
rng = np.random.default_rng(7)

# Create one mechanical measurement feature.
force_newtons = np.linspace(10, 100, 40)

# Create noisy deflection measurements.
deflection_mm = 0.42 * force_newtons + 3.0
noise_mm = rng.normal(0.0, 2.0, force_newtons.size)
deflection_mm = deflection_mm + noise_mm

# Scale the feature for stable comparison.
x_mean = force_newtons.mean()
x_std = force_newtons.std()
x_scaled = (force_newtons - x_mean) / x_std

# Validate the training arrays.
assert x_scaled.shape == deflection_mm.shape
assert x_scaled.size >= 10

# Define mean squared error loss.
def mse_loss(x_values, y_values, slope, intercept):
    predictions = slope * x_values + intercept
    errors = predictions - y_values
    return float(np.mean(errors ** 2))

# Run batch gradient descent silently.
def run_batch_descent(x_values, y_values, rate, steps):
    slope = 0.0
    intercept = 0.0
    losses = []

    for step in range(steps):
        predictions = slope * x_values + intercept
        errors = predictions - y_values
        slope_grad = 2.0 * np.mean(errors * x_values)

        intercept_grad = 2.0 * np.mean(errors)
        slope = slope - rate * slope_grad
        intercept = intercept - rate * intercept_grad
        losses.append(mse_loss(x_values, y_values, slope, intercept))

    return np.array(losses)

# Train with reasonable and excessive rates.
safe_losses = run_batch_descent(x_scaled, deflection_mm, 0.08, 35)
bad_losses = run_batch_descent(x_scaled, deflection_mm, 1.15, 35)

# Detect divergence using simple curve checks.
def divergence_flag(losses):
    finite_values = np.all(np.isfinite(losses))
    growth_ratio = losses[-1] / max(losses[0], 1e-12)
    upward_steps = np.sum(np.diff(losses) > 0)
    return (not finite_values) or growth_ratio > 5 or upward_steps > 20

# Summarize without printing training logs.
print(f"Safe run final loss: {safe_losses[-1]:.2f}")
print(f"Aggressive run final loss: {bad_losses[-1]:.2e}")
print(f"Safe run divergent? {divergence_flag(safe_losses)}")
print(f"Aggressive run divergent? {divergence_flag(bad_losses)}")

# Plot both learning curves together.
plt.figure(figsize=(7, 4))
plt.plot(safe_losses, label="reasonable learning rate")
plt.plot(bad_losses, label="too large learning rate")
plt.yscale("log")
plt.xlabel("Gradient descent iteration")
plt.ylabel("Mean squared error, log scale")
plt.title("Divergence detection from learning curves")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Slow Learning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1783664688" width="250">



>* Error decreases, but only very slowly
>* Updates are cautious and inefficient

>* Tiny learning rates make updates too cautious
>* Unscaled features can slow gradient descent

>* Judge progress against iterations and target error
>* Distinguish slow learning from true convergence



```
#@title Python Code - Slow Learning

# Slow learning means tiny useful updates.
# We compare two learning rates.
# Learning curves reveal training speed.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make the example deterministic.
rng = np.random.default_rng(7)

# Create one mechanical engineering feature.
force_newtons = np.linspace(10, 100, 40)

# Create linear deflection measurements.
noise_mm = rng.normal(0, 0.8, force_newtons.size)
deflection_mm = 0.45 * force_newtons + 3.0 + noise_mm

# Scale the feature for stable gradients.
x_mean = force_newtons.mean()
x_std = force_newtons.std()
x_scaled = (force_newtons - x_mean) / x_std

# Validate matching vector lengths.
assert x_scaled.shape == deflection_mm.shape

# Define mean squared error.
def mse_loss(x_values, y_values, slope, intercept):
    predictions = slope * x_values + intercept
    errors = predictions - y_values
    return float(np.mean(errors ** 2))

# Run batch gradient descent silently.
def batch_descent(x_values, y_values, rate, steps):
    slope, intercept = 0.0, 0.0
    losses = []
    n_points = len(y_values)

    for step in range(steps):
        predictions = slope * x_values + intercept
        errors = predictions - y_values
        slope_grad = 2 * np.mean(errors * x_values)

        intercept_grad = 2 * np.mean(errors)
        slope -= rate * slope_grad
        intercept -= rate * intercept_grad
        losses.append(mse_loss(x_values, y_values, slope, intercept))

    return np.array(losses), slope, intercept

# Compare slow and reasonable learning.
slow_losses, slow_slope, slow_intercept = batch_descent(
    x_scaled, deflection_mm, 0.001, 120
)

# Use a larger safe learning rate.
fast_losses, fast_slope, fast_intercept = batch_descent(
    x_scaled, deflection_mm, 0.08, 120
)

# Summarize progress without printing logs.
slow_drop = slow_losses[0] - slow_losses[-1]
fast_drop = fast_losses[0] - fast_losses[-1]

# Print only key diagnostic values.
print(f"Slow rate final MSE: {slow_losses[-1]:.2f}")
print(f"Reasonable rate final MSE: {fast_losses[-1]:.2f}")
print(f"Slow total MSE drop: {slow_drop:.2f}")
print(f"Reasonable total MSE drop: {fast_drop:.2f}")

# Plot both learning curves together.
plt.figure(figsize=(7, 4))
plt.plot(slow_losses, label="too small rate")
plt.plot(fast_losses, label="reasonable rate")

# Label the diagnostic learning curve.
plt.xlabel("Batch gradient descent iteration")
plt.ylabel("Mean squared error")
plt.title("Slow learning versus reasonable convergence")
plt.legend()

# Display the single required plot.
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
