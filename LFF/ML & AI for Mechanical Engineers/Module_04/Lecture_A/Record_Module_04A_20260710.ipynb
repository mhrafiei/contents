## **1. Regression Intuition**

### **1.1. Core Concept**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1783663665" width="250">



>* Outcomes depend on multiple measurable engineering inputs
>* Features push predictions upward or downward

>* Linear models add feature effects directly
>* Simple patterns support interpretable engineering predictions

>* Predict outcomes by weighting input information
>* Summarize relationships and identify influential features



```
#@title Python Code - Core Concept

# Linear regression combines weighted engineering features.
# This example predicts machine energy use.
# We compute predictions manually using NumPy.

# NumPy and Matplotlib are available in Colab.
import numpy as np
import matplotlib.pyplot as plt

# Store four engineering observations as rows.
features = np.array([
    [10.0, 1200.0, 35.0],
    [15.0, 1500.0, 40.0],
    [20.0, 1800.0, 45.0],
    [25.0, 2100.0, 50.0],
])

# Name each feature for interpretation.
feature_names = ["load_kN", "speed_rpm", "temperature_C"]

# Choose interpretable weights for each feature.
weights = np.array([1.8, 0.015, 0.6])
intercept = 12.0

# Validate shapes before matrix multiplication.
assert features.shape[1] == weights.shape[0]
assert len(feature_names) == weights.shape[0]

# Compute weighted feature contributions manually.
contributions = features * weights
predicted_energy = contributions.sum(axis=1) + intercept

# Create measured values with small residual errors.
measured_energy = np.array([70.0, 86.0, 103.0, 119.0])
residuals = measured_energy - predicted_energy

# Compute mean squared error from residuals.
mse = np.mean(residuals ** 2)

# Print a compact interpretation summary.
print("Linear regression prediction rule:")
print("energy = intercept + weighted feature sum")
print(f"intercept contribution: {intercept:.1f} kWh")

# Print each feature weight clearly.
for name, weight in zip(feature_names, weights):
    print(f"{name} weight: {weight:.3f}")

# Print one prediction and overall error.
print(f"first predicted energy: {predicted_energy[0]:.1f} kWh")
print(f"mean squared error: {mse:.2f}")

# Plot residuals to inspect prediction errors.
plt.figure(figsize=(6, 4))
plt.axhline(0, color="black", linewidth=1)
plt.scatter(predicted_energy, residuals, color="tab:blue", s=70)
plt.xlabel("Predicted energy use (kWh)")
plt.ylabel("Residual: measured minus predicted")
plt.title("Residual Plot for Manual Linear Regression")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Feature Weights**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1783663663" width="250">



>* Weights show each feature’s prediction influence
>* Larger weights mean stronger model impact

>* Weight signs show prediction direction
>* Weight size depends on feature scale

>* Weights learn patterns from observed data
>* Strong weights show association, not causation



```
#@title Python Code - Feature Weights

# Feature weights scale engineering inputs.
# Linear regression adds weighted features.
# Units affect weight interpretation.

# Import NumPy for array calculations.
import numpy as np

# Store three building examples.
features = np.array([
    [120.0, 8.0, 2.0],
    [200.0, 5.0, 6.0],
    [160.0, 12.0, 4.0]
])

# Name each engineering input feature.
feature_names = ["area_m2", "insulation_score", "occupants"]

# Choose illustrative learned feature weights.
weights = np.array([0.35, -4.0, 6.5])

# Add a baseline energy use term.
intercept = 25.0

# Validate compatible feature and weight shapes.
assert features.shape[1] == weights.shape[0]

# Compute weighted contributions for every building.
contributions = features * weights

# Sum contributions and add the intercept.
predictions = contributions.sum(axis=1) + intercept

# Show the model equation compactly.
print("Prediction = 25 + 0.35*area - 4*insulation + 6.5*occupants")

# Show one building contribution breakdown.
print("Building 1 weighted contributions:")

# Print each feature contribution clearly.
for name, value in zip(feature_names, contributions[0]):
    print(f"  {name}: {value:.1f} kWh")

# Show final predictions without large arrays.
print("Predicted energy use, kWh:", np.round(predictions, 1).tolist())

# Highlight sign interpretation for learners.
print("Positive weights increase predictions; negative weights decrease them.")

```

### **1.3. Weighted Feature Sum**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1783663667" width="250">



>* Predictions combine features using learned weights
>* Weights show each feature’s influence

>* Weights turn judgments into numerical predictions
>* Weight signs reveal each feature’s influence

>* Feature scales shape weight meanings
>* Weighted inputs combine into clear predictions



```
#@title Python Code - Weighted Feature Sum

# Weighted sums combine engineering measurements.
# Each feature receives one numerical weight.
# NumPy makes manual prediction transparent.

# Import NumPy for small vector calculations.
import numpy as np

# Store one machine design as features.
feature_names = np.array(["thickness_cm", "temperature_C", "load_kN"])

# Use measured values for one component.
x_values = np.array([2.5, 80.0, 12.0])

# Assign interpretable linear regression weights.
weights = np.array([18.0, -0.25, 3.5])

# Add a baseline prediction intercept.
intercept = 40.0

# Validate matching feature and weight lengths.
assert x_values.shape == weights.shape, "Feature and weight shapes must match."

# Compute each weighted feature contribution.
contributions = x_values * weights

# Sum contributions and add the intercept.
prediction = intercept + np.sum(contributions)

# Print a compact explanation table.
print("Manual linear regression weighted sum")

# Show the baseline intercept first.
print(f"intercept contribution: {intercept:.2f}")

# Display each feature contribution clearly.
for name, value, weight, part in zip(feature_names, x_values, weights, contributions):
    print(f"{name}: {value:.2f} * {weight:.2f} = {part:.2f}")

# Print the final predicted engineering outcome.
print(f"predicted strength index: {prediction:.2f}")

```

## **2. Manual NumPy Predictions**

### **2.1. Feature Weighting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1783663653" width="250">



>* Weights show each feature’s prediction influence
>* Predictions combine many weighted input values

>* Rows store observations; weights match features
>* Weighted sums plus baseline create predictions

>* Weights control each feature’s influence
>* Correct feature preparation prevents misleading predictions



```
#@title Python Code - Feature Weighting

# Manual predictions reveal feature weighting.
# Each feature receives one weight.
# NumPy multiplies matching feature columns.

# Import NumPy for array calculations.
import numpy as np

# Store apartment features by row.
feature_names = np.array([
    "area_m2", "occupants", "temp_c", "electric_heat"
])

# Create three small apartment examples.
X = np.array([
    [55.0, 1.0, 8.0, 0.0],
    [80.0, 3.0, 5.0, 1.0],
    [42.0, 2.0, 12.0, 0.0]
])

# Define one weight per feature.
weights = np.array([2.4, 35.0, -4.0, 120.0])

# Define the baseline monthly usage.
intercept = 90.0

# Check feature and weight alignment.
assert X.shape[1] == weights.shape[0]

# Multiply every feature by its weight.
weighted_features = X * weights

# Add weighted contributions and intercept.
predictions = weighted_features.sum(axis=1) + intercept

# Show the feature order used.
print("Feature order:", feature_names.tolist())

# Show one row of weighted contributions.
print("Apartment 1 contributions:", weighted_features[0].round(1).tolist())

# Show all manual predictions.
print("Predicted kWh:", predictions.round(1).tolist())

# Compare with compact matrix multiplication.
matrix_predictions = X @ weights + intercept

# Confirm both methods match exactly.
print("Matrix method matches:", np.allclose(predictions, matrix_predictions))

```

### **2.2. Matrix Dot Product**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1783663655" width="250">



>* Rows hold cases; columns hold measured features
>* Dot products combine features and weights

>* Match feature columns to weight order
>* Organized arrays make predictions meaningful

>* Dot products scale predictions across datasets
>* Each row gets one repeatable prediction



```
#@title Python Code - Matrix Dot Product

# Matrix dot products combine aligned feature weights.
# Each row represents one engineering observation.
# Each column must match one coefficient.

import numpy as np

# Store three machine operating observations.
features = np.array([
    [1200.0, 8.0, 35.0],
    [1500.0, 9.5, 42.0],
    [1800.0, 11.0, 48.0]
])

# Store matching linear regression weights.
weights = np.array([0.015, 2.0, -0.4])

# Store the intercept separately.
intercept = 12.0

# Validate feature and weight alignment.
assert features.shape[1] == weights.shape[0]

# Compute predictions using matrix multiplication.
predictions = features @ weights + intercept

# Compute one row manually for comparison.
manual_first = np.sum(features[0] * weights) + intercept

# Print compact teaching results.
print("Feature matrix shape:", features.shape)
print("Weight vector shape:", weights.shape)
print("Predictions:", np.round(predictions, 2))
print("First prediction manually:", round(manual_first, 2))
print("First prediction by dot product:", round(predictions[0], 2))

```

### **2.3. NumPy Prediction Arrays**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1783663650" width="250">



>* Batch inputs produce one prediction per row
>* Arrays store model outputs compactly

>* Match feature columns to model weights
>* Keep shapes and preprocessing consistent

>* Compare predictions with actual outcomes
>* Reuse arrays for errors and visualization



```
#@title Python Code - NumPy Prediction Arrays

# Manual predictions use aligned NumPy arrays.
# Each row represents one engineering observation.
# Each column matches one model weight.

# NumPy is available in Google Colab.
import numpy as np

# Store apartment features in consistent column order.
feature_names = np.array(["area_m2", "bedrooms", "distance_km"])

# Rows represent three apartment listings.
X = np.array([
    [55.0, 1.0, 4.0],
    [80.0, 2.0, 7.5],
    [120.0, 3.0, 12.0],
])

# Weights match the feature column order.
weights = np.array([3.2, 18.0, -2.5])

# The intercept shifts every prediction equally.
intercept = 45.0

# Validate shapes before multiplying arrays.
assert X.shape[1] == weights.shape[0]

# Matrix multiplication creates one prediction per row.
predictions = X @ weights + intercept

# Actual prices support quick error checking.
actual_prices = np.array([230.0, 315.0, 430.0])

# Residuals show actual minus predicted values.
residuals = actual_prices - predictions

# Mean squared error summarizes squared residual size.
mse = np.mean(residuals ** 2)

# Print compact shape information for learners.
print("Feature names:", feature_names.tolist())
print("X shape:", X.shape, "weights shape:", weights.shape)
print("Prediction array shape:", predictions.shape)
print("Predicted prices:", np.round(predictions, 1))
print("Residuals:", np.round(residuals, 1))
print("Mean squared error:", round(float(mse), 2))

```

## **3. Regression Error Analysis**

### **3.1. Residual Plots**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1783663657" width="250">



>* Residuals show each prediction error
>* Random scatter suggests a suitable linear model

>* Residual plots reveal hidden model bias
>* Patterns show missed relationships or errors

>* Spot curves, funnels, clusters, and outliers
>* Check whether errors are truly random



```
#@title Python Code - Residual Plots

# Residual plots reveal regression error patterns.
# NumPy computes predictions and residuals manually.
# Matplotlib shows whether errors look random.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Create deterministic engineering-style input data.
np.random.seed(7)
area_sqft = np.linspace(700, 2600, 30)
noise = np.random.normal(0, 18000, area_sqft.size)

# Define actual prices with slight curvature.
actual_price = 65000 + 155 * area_sqft + 0.018 * area_sqft**2 + noise

# Build a simple feature matrix manually.
ones_column = np.ones(area_sqft.size)
X = np.column_stack((ones_column, area_sqft))

# Validate the matrix and target sizes.
assert X.shape[0] == actual_price.shape[0]
assert X.shape[1] == 2

# Solve linear regression weights using NumPy.
weights = np.linalg.inv(X.T @ X) @ X.T @ actual_price
predicted_price = X @ weights

# Compute residuals and mean squared error.
residuals = actual_price - predicted_price
mse = np.mean(residuals**2)

# Print a compact numerical summary.
print(f"Intercept: ${weights[0]:,.0f}")
print(f"Slope: ${weights[1]:,.2f} per square foot")
print(f"Mean squared error: {mse:,.0f}")
print("Residuals should scatter randomly around zero.")

# Plot residuals against predicted values.
plt.figure(figsize=(7, 4))
plt.scatter(predicted_price, residuals, color="steelblue", edgecolor="black")
plt.axhline(0, color="red", linestyle="--", linewidth=2)

# Label the residual plot clearly.
plt.xlabel("Predicted house price, dollars")
plt.ylabel("Residual, actual minus predicted")
plt.title("Residual Plot for Manual Linear Regression")
plt.grid(True, alpha=0.3)

# Display the single teaching plot.
plt.show()

```

### **3.2. Mean Squared Error**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1783663659" width="250">



>* MSE summarizes prediction errors using residuals
>* Squaring prevents cancellation; lower MSE is better

>* Large errors strongly increase MSE
>* Highlights costly, unusually poor predictions

>* MSE uses squared, less intuitive units
>* Compare with plots and domain context



```
#@title Python Code - Mean Squared Error

# This example computes mean squared error.
# Residuals compare predictions with measured outcomes.
# Squaring residuals emphasizes larger engineering mistakes.

# NumPy and Matplotlib are available in Colab.
import numpy as np
import matplotlib.pyplot as plt

# Store measured beam deflections in millimeters.
actual_mm = np.array([2.0, 2.8, 3.6, 4.5, 5.1, 6.0])

# Store model predictions from manual linear regression.
predicted_mm = np.array([2.1, 2.6, 3.9, 4.2, 5.4, 5.7])

# Validate matching shapes before subtracting arrays.
assert actual_mm.shape == predicted_mm.shape, "Shapes must match"

# Compute residuals as actual minus predicted values.
residuals_mm = actual_mm - predicted_mm

# Square residuals so signs cannot cancel.
squared_errors = residuals_mm ** 2

# Average squared errors to get MSE.
mse_value = np.mean(squared_errors)

# Print a compact numerical summary.
print(f"Number of observations: {actual_mm.size}")
print(f"Mean squared error: {mse_value:.3f} squared mm")
print(f"Largest absolute residual: {np.max(np.abs(residuals_mm)):.3f} mm")

# Plot residuals against predicted deflection values.
plt.figure(figsize=(6, 4))
plt.scatter(predicted_mm, residuals_mm, color="navy", s=70)
plt.axhline(0, color="black", linewidth=1)

# Label the residual plot clearly.
plt.xlabel("Predicted deflection (mm)")
plt.ylabel("Residual: actual - predicted (mm)")
plt.title("Residual Plot for Beam Deflection Model")

# Add a light grid for readability.
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### **3.3. Prediction Error Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1783663661" width="250">



>* Random errors suggest a suitable model
>* Patterns reveal missing features or bias

>* Errors often grow at data extremes
>* Wider error spread hides unreliable predictions

>* Error patterns reveal bias across conditions
>* Use patterns to improve model reliability



```
#@title Python Code - Prediction Error Patterns

# This example studies regression error patterns.
# Residual plots reveal missed model structure.
# NumPy computes predictions and squared errors.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Make the random example reproducible.
rng = np.random.default_rng(7)

# Create vehicle weights in pounds.
weight_lb = np.linspace(2200, 6200, 60)

# Create curved fuel use with noise.
noise = rng.normal(0, 1.8, weight_lb.size)
true_mpg = 48 - 0.0045 * weight_lb + 0.00000035 * (weight_lb - 4200) ** 2
observed_mpg = true_mpg + noise

# Validate matching one dimensional arrays.
assert weight_lb.ndim == observed_mpg.ndim == 1
assert weight_lb.size == observed_mpg.size

# Build a manual linear regression design matrix.
ones_column = np.ones_like(weight_lb)
X = np.column_stack((ones_column, weight_lb))

# Solve weights using the normal equation.
beta = np.linalg.solve(X.T @ X, X.T @ observed_mpg)
predicted_mpg = X @ beta

# Compute residuals and mean squared error.
residuals = observed_mpg - predicted_mpg
mse = np.mean(residuals ** 2)

# Summarize the fitted weighted combination.
print(f"Prediction = {beta[0]:.2f} + {beta[1]:.5f} * weight_lb")
print(f"Mean squared error: {mse:.2f} mpg^2")
print("Pattern check: curved residuals suggest missing nonlinear behavior.")

# Plot residuals against predicted values.
plt.figure(figsize=(7, 4))
plt.scatter(predicted_mpg, residuals, color="steelblue", alpha=0.8)
plt.axhline(0, color="black", linewidth=1)
plt.xlabel("Predicted fuel economy, mpg")
plt.ylabel("Residual, observed minus predicted")
plt.title("Prediction Error Patterns in Linear Regression")
plt.grid(True, alpha=0.3)
plt.show()

```
