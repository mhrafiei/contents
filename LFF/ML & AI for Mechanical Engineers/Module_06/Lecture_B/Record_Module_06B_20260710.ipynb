## **1. Unsupervised Pattern Discovery**

### **1.1. Unlabeled Data Patterns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_01_01.jpg?v=1783669637" width="250">



>* Engineering measurements often lack predefined labels
>* Clustering reveals natural patterns for interpretation

>* Clustering groups observations by measured similarity
>* Engineers interpret centroids using system context

>* Clustering reveals hidden engineering operating modes
>* Interpret clusters using domain knowledge



```
#@title Python Code - Unlabeled Data Patterns

# Clustering can reveal hidden engineering operating patterns.
# Unlabeled measurements still contain useful similarity structure.
# This example uses synthetic pump sensor data.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Make random values repeatable for teaching.
rng = np.random.default_rng(7)

# Define hidden operating regimes for simulation.
centers = np.array([[35, 1.2], [55, 2.0], [75, 3.4]])

# Create unlabeled pressure and vibration measurements.
points = np.vstack([rng.normal(c, [3.0, 0.18], size=(18, 2)) for c in centers])

# Validate the small teaching dataset shape.
assert points.shape == (54, 2)

# Choose three starting centroids from observations.
centroids = points[[2, 24, 43]].copy()

# Run a few transparent k-means iterations.
for step in range(6):
    distances = np.linalg.norm(points[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([points[labels == k].mean(axis=0) for k in range(3)])

# Count observations assigned to each cluster.
counts = np.bincount(labels, minlength=3)

# Print concise engineering interpretation.
print("Unlabeled pump records:", len(points))
print("Cluster sizes:", counts.tolist())
print("Centroids are pressure psi and vibration in/s.")
print(np.round(centroids, 2))
print("Meaning requires engineering judgment, not labels.")

# Plot discovered patterns in feature space.
plt.figure(figsize=(7, 5))

# Show records colored by discovered cluster.
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap="viridis", s=55)

# Mark centroids as representative operating patterns.
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)

# Label axes using mechanical engineering units.
plt.xlabel("Discharge pressure (psi)")
plt.ylabel("Vibration velocity (in/s)")
plt.title("K-means finds patterns without condition labels")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Hidden Data Groups**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_01_02.jpg?v=1783669635" width="250">

```
#@title Python Code - Hidden Data Groups

# Hidden groups can appear without labels.
# Mechanical measurements may cluster by behavior.
# This example uses simple k-means steps.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Make the random example reproducible.
rng = np.random.default_rng(7)

# Create three unlabeled pump behavior groups.
normal = rng.normal([80, 2.0], [4, 0.15], (12, 2))
hot = rng.normal([95, 3.1], [4, 0.18], (12, 2))
vibrating = rng.normal([78, 4.0], [3, 0.20], (12, 2))

# Combine observations without keeping labels.
X = np.vstack([normal, hot, vibrating])
assert X.shape == (36, 2)

# Choose three starting centroids from observations.
centroids = X[[0, 14, 28]].copy()
assert centroids.shape == (3, 2)

# Repeat assignment and centroid update steps.
for step in range(6):
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    groups = np.argmin(distances, axis=1)
    centroids = np.array([X[groups == k].mean(axis=0) for k in range(3)])

# Count observations assigned to each hidden group.
counts = np.bincount(groups, minlength=3)

# Print a compact engineering interpretation.
print("Unlabeled pump records:", len(X))
print("Hidden group sizes:", counts.tolist())
print("Centroids: flow gpm, vibration mm/s")
for k, center in enumerate(centroids, start=1):
    print(f"Group {k}: {center[0]:.1f} gpm, {center[1]:.2f} mm/s")

# Plot observations and discovered centroids.
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=groups, cmap="viridis", s=55)
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)
plt.xlabel("Flow rate, gallons per minute")
plt.ylabel("Vibration level, millimeters per second")
plt.title("K-means reveals hidden pump behavior groups")
plt.grid(True, alpha=0.3)
plt.show()

```

```
#@title Python Code - Engineering Data Patterns

# Clustering reveals hidden engineering operating patterns.
# Synthetic sensor data keeps this example simple.
# K-means is implemented without machine learning libraries.

# Import numerical and plotting libraries.
import numpy as np
import matplotlib.pyplot as plt

# Make random data reproducible.
rng = np.random.default_rng(7)

# Define three unlabeled operating regimes.
centers = np.array([[2.0, 35.0], [6.0, 55.0], [9.0, 80.0]])

# Generate vibration and temperature measurements.
points = np.vstack([rng.normal(c, [0.45, 3.0], size=(18, 2)) for c in centers])

# Validate the small teaching dataset.
assert points.shape == (54, 2)

# Choose initial centroids from observations.
centroids = points[[0, 20, 40]].copy()

# Repeat assignment and update steps.
for step in range(6):
    distances = np.linalg.norm(points[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    centroids = np.array([points[labels == k].mean(axis=0) for k in range(3)])

# Count observations in each discovered group.
counts = np.bincount(labels, minlength=3)

# Print concise engineering interpretation.
print("Unlabeled vibration-temperature observations were grouped by similarity.")
print(f"Cluster sizes: {counts.tolist()} observations.")
print("Centroids show typical vibration inches/sec and temperature Celsius.")

# Print each centroid compactly.
for k, center in enumerate(centroids):
    print(f"Cluster {k}: vibration={center[0]:.2f}, temperature={center[1]:.1f}")

# Create one plot showing discovered patterns.
plt.figure(figsize=(7, 4.5))

# Plot observations colored by assigned cluster.
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap="viridis", s=55)

# Plot centroids as engineering summaries.
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=180)

# Label axes with engineering units.
plt.xlabel("Vibration amplitude (inches/sec)")
plt.ylabel("Bearing temperature (°C)")

# Add a clear teaching title.
plt.title("K-means finds unlabeled engineering data patterns")
plt.grid(True, alpha=0.3)
plt.show()

```

## **2. K Means Steps**

### **2.1. Initialize Centroids**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_02_01.jpg?v=1783669629" width="250">



>* Centroids start as tentative cluster centers
>* Starting positions affect convergence and clusters

>* Randomly choose data points as starting centroids
>* Simple method, but results can vary

>* Choose starting centroids carefully
>* Rerun or spread starts to improve clusters



```
#@title Python Code - Initialize Centroids

# Demonstrate random centroid initialization.
# Use small mechanical measurement data.
# Keep output short and visual.

import numpy as np
import matplotlib.pyplot as plt

# Create repeatable synthetic vibration data.
rng = np.random.default_rng(7)

# Store speed and vibration measurements.
data = np.array([
    [1180, 0.18], [1210, 0.22], [1195, 0.20],

    [1780, 0.55], [1815, 0.62], [1765, 0.58],
    [2380, 1.05], [2420, 1.12], [2395, 1.08]
])

# Choose the desired cluster count.
k_clusters = 3

# Validate enough observations exist.
if data.shape[0] < k_clusters:
    raise ValueError("Need at least k data points.")

# Randomly choose existing rows as centroids.
chosen_rows = rng.choice(data.shape[0], k_clusters, replace=False)
centroids = data[chosen_rows].copy()

# Print only compact teaching output.
print("Initial centroid row indices:", chosen_rows.tolist())
print("Centroids are existing measurements.")
print(np.round(centroids, 2))

# Plot data and initialized centroids.
plt.figure(figsize=(6, 4))
plt.scatter(data[:, 0], data[:, 1], label="measurements")
plt.scatter(centroids[:, 0], centroids[:, 1], marker="X", s=180, label="initial centroids")
plt.xlabel("shaft speed, rpm")
plt.ylabel("vibration amplitude, mm/s")
plt.title("K-means starts by selecting initial centroids")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Centroid Update**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_02_02.jpg?v=1783669631" width="250">



>* Move centroids to assigned points’ average position
>* Repeated updates reveal typical cluster profiles

>* Recalculate centroids from current cluster members
>* Repeat assignments until clusters stabilize

>* Average assigned points to update each centroid
>* Scale features and handle empty clusters



```
#@title Python Code - Centroid Update

# Centroid updates summarize assigned engineering measurements.
# This example uses tiny pump operating data.
# We update centroids from scratch using NumPy.

import numpy as np
import matplotlib.pyplot as plt

# Create small pump data: flow and vibration.
flow_gpm = np.array([48, 52, 50, 82, 86, 84], dtype=float)
vibration_mm_s = np.array([1.8, 2.1, 1.9, 5.8, 6.2, 6.0], dtype=float)

# Stack features into one data matrix.
X = np.column_stack((flow_gpm, vibration_mm_s))
assert X.shape == (6, 2), "Expected six pumps and two features"

# Choose two rough starting centroids.
centroids = np.array([[49.0, 2.0], [83.0, 5.7]], dtype=float)
assert centroids.shape == (2, 2), "Expected two centroids"

# Compute distances from every pump to every centroid.
differences = X[:, np.newaxis, :] - centroids[np.newaxis, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))
labels = np.argmin(distances, axis=1)

# Update each centroid using assigned pump averages.
new_centroids = centroids.copy()
for cluster_id in range(centroids.shape[0]):
    cluster_points = X[labels == cluster_id]
    if cluster_points.shape[0] > 0:
        new_centroids[cluster_id] = cluster_points.mean(axis=0)

# Print compact centroid interpretation for engineers.
print("Assignments:", labels.tolist())
print("Old centroids:", np.round(centroids, 2).tolist())
print("Updated centroids:", np.round(new_centroids, 2).tolist())
print("Centroid 0 means typical low-vibration pump behavior.")
print("Centroid 1 means typical high-vibration pump behavior.")

# Plot pumps and centroid movement.
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=80, cmap="viridis")
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=140, c="red")
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], marker="*", s=180, c="black")
plt.xlabel("Flow rate, gallons per minute")
plt.ylabel("Vibration, millimeters per second")
plt.title("K-Means Centroid Update From Assigned Pumps")
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Centroid Update**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_02_03.jpg?v=1783669633" width="250">



>* Move centroids to assigned points' average
>* Repeated updates improve cluster representatives

>* Recompute each cluster’s central machine profile
>* Repeat assignments as centroids shift

>* Average assigned points across every feature
>* Handle empty clusters during centroid updates



```
#@title Python Code - Centroid Update

# Demonstrate centroid updates from scratch.
# Use small mechanical sensor measurements.
# Keep output concise and visual.

import numpy as np
import matplotlib.pyplot as plt

# Store vibration and temperature measurements.
points = np.array([
    [0.18, 42.0], [0.22, 44.0],
    [0.25, 43.0], [0.80, 78.0],
    [0.85, 82.0], [0.90, 80.0]
])

# Start with two rough centroids.
centroids = np.array([
    [0.20, 41.0], [0.75, 76.0]
])

# Validate the feature dimensions.
assert points.shape[1] == centroids.shape[1]

# Compute distances to every centroid.
differences = points[:, None, :] - centroids[None, :, :]
distances = np.sqrt(np.sum(differences ** 2, axis=2))
assignments = np.argmin(distances, axis=1)

# Update each centroid by averaging members.
new_centroids = centroids.copy()
for cluster_id in range(len(centroids)):
    members = points[assignments == cluster_id]
    if len(members) > 0:
        new_centroids[cluster_id] = members.mean(axis=0)

# Print compact engineering interpretation.
print("Old centroids:", np.round(centroids, 2).tolist())
print("New centroids:", np.round(new_centroids, 2).tolist())
print("Assignments:", assignments.tolist())
print("Centroids moved toward current cluster averages.")

# Plot points and centroid movement.
colors = np.array(["tab:blue", "tab:orange"])
plt.figure(figsize=(7, 4))
plt.scatter(points[:, 0], points[:, 1], c=colors[assignments], s=80)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=140, c="black")
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], marker="*", s=220, c="red")

# Label axes using engineering units.
plt.xlabel("Vibration amplitude, inches per second")
plt.ylabel("Operating temperature, degrees Celsius")
plt.title("Centroid update after one assignment step")
plt.grid(True, alpha=0.3)
plt.show()

```

## **3. Cluster Review**

### **3.1. Centroid Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_03_01.jpg?v=1783669627" width="250">



>* Centroids show typical cluster behavior
>* Compare patterns to real engineering conditions

>* Translate scaled centroids into real engineering values
>* Use system context to explain cluster meaning

>* Clear centroids suggest useful, actionable clusters
>* Use plots; centroids guide hypotheses, not causation



```
#@title Python Code - Centroid Interpretation

# Interpret centroids using engineering feature context.
# Compare cluster centers against overall averages.
# Use one plot and concise printed summaries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create deterministic mechanical operating data.
rng = np.random.default_rng(7)
cluster_sizes = [18, 18, 18]

# Define realistic centroid-like operating states.
true_centers = np.array([
    [0.18, 1800.0, 62.0, 35.0],
    [0.42, 2200.0, 78.0, 70.0],
    [0.75, 1600.0, 92.0, 85.0],
])

# Define feature names with engineering units.
features = [
    "vibration_mm_s", "speed_rpm",
    "temperature_C", "load_percent",
]

# Generate compact clusters around each state.
noise_scale = np.array([0.05, 90.0, 4.0, 6.0])
blocks = []

for center, size in zip(true_centers, cluster_sizes):
    block = rng.normal(center, noise_scale, size=(size, 4))
    blocks.append(block)

# Combine observations into one small dataset.
X = np.vstack(blocks)
assert X.shape == (54, 4)

# Standardize features before k-means distance calculations.
means = X.mean(axis=0)
stds = X.std(axis=0)
Z = (X - means) / stds

# Initialize centroids from three representative observations.
centroids = Z[[0, 20, 40]].copy()
labels = np.zeros(Z.shape[0], dtype=int)

# Run simple k-means assignment and update steps.
for step in range(8):
    distances = ((Z[:, None, :] - centroids[None, :, :]) ** 2).sum(axis=2)
    labels = distances.argmin(axis=1)
    centroids = np.array([Z[labels == k].mean(axis=0) for k in range(3)])

# Convert standardized centroids back to physical units.
centers_units = centroids * stds + means
centroid_table = pd.DataFrame(centers_units, columns=features)

# Add compact engineering interpretation labels.
centroid_table["interpretation"] = [
    "normal efficient operation",
    "heavy warm operation",
    "high-risk thermal vibration state",
]

# Print concise centroid interpretation summary.
print("Centroids translated back to engineering units:")
for idx, row in centroid_table.iterrows():
    text = f"Cluster {idx}: {row['interpretation']}"
    print(text)
    values = f"  vib={row['vibration_mm_s']:.2f}, temp={row['temperature_C']:.1f}, load={row['load_percent']:.0f}"
    print(values)

# Plot clusters and centroids for visual review.
plt.figure(figsize=(7, 5))
scatter = plt.scatter(X[:, 0], X[:, 2], c=labels, cmap="viridis", s=45)
plt.scatter(centers_units[:, 0], centers_units[:, 2], c="red", marker="X", s=180)
plt.xlabel("Vibration amplitude (mm/s)")
plt.ylabel("Bearing temperature (C)")
plt.title("Centroid interpretation: typical operating states")
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.2. Centroid Interpretation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/ML & AI for Mechanical Engineers/Module_06/Lecture_B/image_03_02.jpg?v=1783669620" width="250">



>* Centroids show typical cluster behavior
>* Translate averages into engineering meaning

>* Check feature scales and engineering units
>* Compare centroids to real operating regimes

>* Link centroids to engineering actions
>* Validate with plots and judgment



```
#@title Python Code - Centroid Interpretation

# Interpret centroids using engineering feature meanings.
# This example uses small pump operating data.
# K-means is implemented without machine learning libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create deterministic pump measurements with engineering units.
np.random.seed(7)
flow = np.array([48, 50, 52, 80, 83, 85, 60, 62, 64], dtype=float)
power = np.array([9.5, 9.8, 10.1, 18.5, 19.0, 19.4, 16.0, 16.5, 17.0], dtype=float)
vibration = np.array([0.08, 0.09, 0.10, 0.12, 0.13, 0.14, 0.32, 0.35, 0.38], dtype=float)

# Combine features into one data matrix.
X = np.column_stack([flow, power, vibration])
assert X.shape == (9, 3)

# Standardize features before distance calculations.
means = X.mean(axis=0)
stds = X.std(axis=0)
X_scaled = (X - means) / stds

# Choose three starting centroids from actual observations.
centroids = X_scaled[[0, 3, 6]].copy()
labels = np.zeros(X_scaled.shape[0], dtype=int)

# Repeat assignment and centroid update steps.
for step in range(6):
    distances = np.linalg.norm(X_scaled[:, None, :] - centroids[None, :, :], axis=2)
    labels = distances.argmin(axis=1)
    for cluster_id in range(3):
        members = X_scaled[labels == cluster_id]
        if len(members) > 0:
            centroids[cluster_id] = members.mean(axis=0)

# Convert centroids back to engineering units.
centroids_units = centroids * stds + means
summary = pd.DataFrame(centroids_units, columns=["flow_gpm", "power_kW", "vibration_in_s"])
summary["count"] = np.bincount(labels, minlength=3)

# Print concise centroid interpretations.
print("Centroid interpretation for pump clusters:")
for cluster_id, row in summary.iterrows():
    if row["vibration_in_s"] > 0.25:
        meaning = "mechanically unstable behavior"
    elif row["power_kW"] > 15 and row["flow_gpm"] > 75:
        meaning = "high-output, high-power operation"
    else:
        meaning = "efficient low-load operation"
    print(f"Cluster {cluster_id}: {row['count']:.0f} pumps, {meaning}.")

# Plot observations and centroids for engineering review.
colors = np.array(["tab:blue", "tab:orange", "tab:green"])
plt.figure(figsize=(7, 5))
plt.scatter(flow, power, c=colors[labels], s=80, label="pumps")
plt.scatter(summary["flow_gpm"], summary["power_kW"], c="black", marker="X", s=180, label="centroids")
plt.xlabel("Flow rate, gallons per minute")
plt.ylabel("Power consumption, kW")
plt.title("Centroids summarize typical pump operating regimes")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **3.3. Centroid Interpretation**

```
#@title Python Code - Centroid Interpretation

# Interpret centroids using engineering feature context.
# Synthetic pump data keeps clustering transparent.
# Plots connect numerical centers to meaning.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set deterministic randomness for repeatable results.
rng = np.random.default_rng(7)

# Define engineering operating prototypes as centroids.
true_centers = np.array([
    [35.0, 18.0, 72.0],
    [62.0, 31.0, 84.0],
    [88.0, 46.0, 68.0],
])

# Create small noisy observations around prototypes.
points_per_group = 18
noise_scale = np.array([4.0, 2.5, 3.0])

# Build unlabeled pump operating measurements.
blocks = []
for center in true_centers:
    block = rng.normal(center, noise_scale, (points_per_group, 3))
    blocks.append(block)

# Combine blocks into one unlabeled dataset.
X = np.vstack(blocks)
assert X.shape == (54, 3)

# Scale features before distance calculations.
means = X.mean(axis=0)
stds = X.std(axis=0)
X_scaled = (X - means) / stds

# Initialize three centroids from fixed observations.
k = 3
centroids = X_scaled[[0, 20, 40]].copy()

# Run simple k-means assignment and update steps.
for step in range(8):
    distances = ((X_scaled[:, None, :] - centroids[None, :, :]) ** 2).sum(axis=2)
    labels = distances.argmin(axis=1)
    centroids = np.array([X_scaled[labels == j].mean(axis=0) for j in range(k)])

# Convert centroids back to engineering units.
centroids_units = centroids * stds + means
columns = ["Flow_gpm", "Pressure_psi", "Efficiency_percent"]
centroid_table = pd.DataFrame(centroids_units, columns=columns)

# Sort clusters by flow for easier interpretation.
centroid_table = centroid_table.sort_values("Flow_gpm").reset_index(drop=True)

# Print compact centroid interpretation summary.
print("Centroid interpretation for pump operation:")
for i, row in centroid_table.iterrows():
    flow = row["Flow_gpm"]
    pressure = row["Pressure_psi"]
    efficiency = row["Efficiency_percent"]
    print(f"Cluster {i}: {flow:.1f} gpm, {pressure:.1f} psi, {efficiency:.1f}% efficiency")

# Add simple engineering labels from centroid patterns.
print("Meaning: low-load, efficient mid-load, and high-flow lower-efficiency regions.")

# Plot observations and interpreted centroids.
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 2], c=labels, cmap="viridis", alpha=0.75)
plt.scatter(centroids_units[:, 0], centroids_units[:, 2], c="red", marker="X", s=180)
plt.xlabel("Flow rate (gpm)")
plt.ylabel("Efficiency (%)")
plt.title("Centroids as representative pump operating profiles")
plt.grid(True, alpha=0.3)
plt.show()

```
