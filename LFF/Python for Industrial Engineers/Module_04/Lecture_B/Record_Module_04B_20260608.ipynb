## **1. Inventory Logic**

### **1.1. Reorder Point**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1780943339" width="250">



>* Order before stock runs out
>* Use demand, delay, and stockout risk

>* Reorder using demand during lead time
>* Python conditions trigger orders at thresholds

>* Higher demand or delays raise reorder points
>* Balance service levels against inventory costs



```
#@title Python Code - Reorder Point

# Reorder points signal timely replenishment decisions.
# Lead time demand drives inventory thresholds.
# Python conditions can trigger simple orders.

# Define a reusable reorder point function.
def reorder_point(daily_demand, lead_time_days, safety_stock):
    lead_time_demand = daily_demand * lead_time_days
    reorder_threshold = lead_time_demand + safety_stock

    return reorder_threshold

# Store a workshop inventory scenario.
item_name = "pump filters"
daily_demand = 18
lead_time_days = 6

safety_stock = 25
current_inventory = 120
order_quantity = 200

# Calculate the reorder threshold.
threshold = reorder_point(
    daily_demand,
    lead_time_days,
    safety_stock,
)

# Decide whether an order is needed.
if current_inventory <= threshold:
    decision = "ORDER NOW"
else:
    decision = "WAIT"

# Compare alternative supplier lead times.
local_supplier = reorder_point(18, 2, 25)
import_supplier = reorder_point(18, 9, 25)

# Print a compact engineering summary.
print(f"Item analyzed: {item_name}")
print(f"Current inventory: {current_inventory} units")
print(f"Reorder point: {threshold} units")
print(f"Decision: {decision}")
print(f"Recommended order quantity: {order_quantity} units")
print(f"Local supplier reorder point: {local_supplier} units")
print(f"Import supplier reorder point: {import_supplier} units")
print("Longer lead time requires earlier reordering.")

```

### **1.2. Safety Stock**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1780943335" width="250">



>* Extra inventory protects against uncertain demand
>* Buffers surprises until replenishment arrives

>* Balance stockout risk against holding costs
>* Use fixed or conditional safety stock rules

>* Safety stock triggers earlier reordering
>* Frequent shortages reveal deeper system problems



```
#@title Python Code - Safety Stock

# Safety stock buffers demand uncertainty.
# Reorder points include expected demand.
# Conditional logic can compare risk levels.

# No extra packages are required.

# Define average demand and lead time.
average_daily_demand = 40
lead_time_days = 5

# Define fixed safety stock scenarios.
safety_stock_options = {
    "low risk": 20,
    "moderate risk": 60,
    "high risk": 100,
}

# Create a reusable reorder calculation.
def calculate_reorder_point(demand_per_day, lead_time, safety_stock):
    expected_lead_time_demand = demand_per_day * lead_time
    reorder_point = expected_lead_time_demand + safety_stock

    return reorder_point

# Classify each buffer using conditions.
def classify_buffer(safety_stock):
    if safety_stock < 40:
        return "lower holding cost, higher stockout risk"

    if safety_stock <= 80:
        return "balanced buffer for typical uncertainty"

    return "higher availability, higher holding cost"

# Print the base planning assumptions.
print("Average demand:", average_daily_demand, "units per day")
print("Supplier lead time:", lead_time_days, "days")
print("Expected lead-time demand:", average_daily_demand * lead_time_days, "units")

# Compare reorder points across scenarios.
for scenario_name, safety_stock in safety_stock_options.items():
    reorder_point = calculate_reorder_point(
        average_daily_demand,
        lead_time_days,
        safety_stock,
    )

    risk_message = classify_buffer(safety_stock)
    print(scenario_name, "safety stock:", safety_stock, "units")
    print("reorder at", reorder_point, "units:", risk_message)

# Show one decision rule outcome.
chosen_safety_stock = safety_stock_options["moderate risk"]
chosen_reorder_point = calculate_reorder_point(
    average_daily_demand,
    lead_time_days,
    chosen_safety_stock,
)
print("Chosen reorder point:", chosen_reorder_point, "units")

```

### **1.3. Reorder Point**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1780943337" width="250">



>* Inventory threshold for placing new orders
>* Covers demand during replenishment delay

>* Use demand and lead time inputs
>* Reorder decisions replace inventory guesswork

>* Balance stockout risk with inventory costs
>* State assumptions before comparing reorder scenarios



```
#@title Python Code - Reorder Point

# Reorder points convert demand into action.
# This example uses simple inventory logic.
# Conditions decide whether ordering is needed.

# Define daily demand in units.
daily_demand_units = 18

# Define supplier lead time in days.
lead_time_days = 4

# Define extra protection against uncertainty.
safety_stock_units = 20

# Define current shelf inventory units.
current_stock_units = 85

# Create a reusable reorder point function.
def calculate_reorder_point(demand_per_day, lead_time, safety_stock):

    # Estimate demand during the waiting period.
    lead_time_demand = demand_per_day * lead_time

    # Add safety stock for uncertainty.
    reorder_point = lead_time_demand + safety_stock

    # Return the threshold inventory level.
    return reorder_point

# Calculate the reorder point threshold.
reorder_point_units = calculate_reorder_point(
    daily_demand_units, lead_time_days, safety_stock_units
)

# Decide whether a new order is needed.
if current_stock_units <= reorder_point_units:

    # Store the recommended inventory action.
    decision_text = "Place a new order now."

else:

    # Store the alternative inventory action.
    decision_text = "Do not reorder yet."

# Compare one lower-stock scenario too.
scenario_stock_units = 70

# Check the second scenario with same rule.
scenario_needs_order = scenario_stock_units <= reorder_point_units

# Print compact teaching results.
print("Daily demand:", daily_demand_units, "units per day")
print("Lead time:", lead_time_days, "days")
print("Safety stock:", safety_stock_units, "units")
print("Reorder point:", reorder_point_units, "units")
print("Current stock:", current_stock_units, "units")
print("Decision:", decision_text)
print("If stock is", scenario_stock_units, "units, reorder:", scenario_needs_order)

```

## **2. Queue Measures**

### **2.1. Arrival Rates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1780943346" width="250">



>* Arrival rate shows demand entering over time
>* High arrivals versus capacity create waiting

>* Average rates can hide demand variation
>* Use peak estimates when arrivals vary

>* Random arrival bursts can create queues
>* Use arrival rates to estimate system performance



```
#@title Python Code - Arrival Rates

# Estimate arrival rates from observed arrivals.
# Compare average and peak demand periods.
# Keep calculations simple and transparent.

# Store observed arrivals for four periods.
period_names = ["Morning", "Midday", "Afternoon", "Evening"]
arrivals = [180, 90, 120, 210]
hours_observed = [2, 3, 3, 2]

# Validate matching list lengths before calculating.
if len(arrivals) != len(hours_observed):
    raise ValueError("Arrival and time lists must match.")

# Calculate arrivals per hour for each period.
arrival_rates = []
for count, hours in zip(arrivals, hours_observed):
    arrival_rates.append(count / hours)

# Calculate the full day average arrival rate.
total_arrivals = sum(arrivals)
total_hours = sum(hours_observed)
average_rate = total_arrivals / total_hours

# Identify the period with the highest rate.
peak_index = arrival_rates.index(max(arrival_rates))
peak_name = period_names[peak_index]
peak_rate = arrival_rates[peak_index]

# Estimate utilization using a stated service capacity.
service_capacity = 100
average_utilization = average_rate / service_capacity
peak_utilization = peak_rate / service_capacity

# Print compact teaching results.
print(f"Total arrivals observed: {total_arrivals} units")
print(f"Total observation time: {total_hours} hours")
print(f"Average arrival rate: {average_rate:.1f} units/hour")
print(f"Peak period: {peak_name} at {peak_rate:.1f} units/hour")
print(f"Capacity assumption: {service_capacity} units/hour")
print(f"Average utilization: {average_utilization:.2f}")
print(f"Peak utilization: {peak_utilization:.2f}")
print("Peak demand may hide behind a reasonable daily average.")

```

### **2.2. Service Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780943350" width="250">



>* Service rate measures completed work per time
>* Compare service capacity with arriving demand

>* Include all tasks in service estimates
>* Ignoring support work understates queue delays

>* Capacity depends on resources and bottlenecks
>* State assumptions when comparing service scenarios



```
#@title Python Code - Service Rate

# Estimate service rate from completed jobs.
# Compare capacity with observed demand.
# Keep assumptions visible for decisions.

# Store observed completed inspections and hours.
completed_jobs = 30
observed_hours = 1.0

# Calculate service rate in jobs per hour.
service_rate = completed_jobs / observed_hours
arrival_rate = 24

# Calculate utilization as demand divided by capacity.
utilization = arrival_rate / service_rate
safety_gap = service_rate - arrival_rate

# Classify the queue risk using utilization.
if utilization < 0.85:
    status = "manageable under this assumption"
else:
    status = "needs attention or added capacity"

# Compare one server with two parallel servers.
servers = 2
combined_rate = servers * service_rate

# Print a concise engineering summary.
print("Service rate:", service_rate, "jobs/hour")
print("Arrival rate:", arrival_rate, "jobs/hour")
print("Utilization:", round(utilization, 2))
print("Capacity gap:", safety_gap, "jobs/hour")
print("Queue status:", status)
print("Two-server capacity:", combined_rate, "jobs/hour")

```

### **2.3. Service Time**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1780943348" width="250">



>* Service time is active work per unit
>* Separate waiting from service to diagnose queues

>* Estimate service time from real observations
>* Variation can create temporary queue backlogs

>* Service time shows time needed per unit
>* State assumptions when comparing process improvements



```
#@title Python Code - Service Time

# Estimate service time from observed jobs.
# Compare scenarios using simple queue measures.
# Keep calculations transparent for beginners.

# Store observed service times in minutes.
service_times = [4.2, 5.1, 3.8, 6.0, 4.9, 5.5]

# Count observations before calculating the average.
observation_count = len(service_times)

# Validate that observations exist before analysis.
if observation_count == 0:
    raise ValueError("Service time list cannot be empty")

# Compute the average service time.
average_service_time = sum(service_times) / observation_count

# Convert average service time into service rate.
service_rate_per_hour = 60 / average_service_time

# Define an expected arrival rate.
arrival_rate_per_hour = 9

# Estimate server utilization under simple assumptions.
utilization = arrival_rate_per_hour / service_rate_per_hour

# Compare with an improved service process.
improved_service_time = average_service_time - 0.8

# Estimate improved service rate and utilization.
improved_rate_per_hour = 60 / improved_service_time
improved_utilization = arrival_rate_per_hour / improved_rate_per_hour

# Show concise engineering results.
print(f"Observed jobs: {observation_count}")
print(f"Average service time: {average_service_time:.2f} minutes/job")
print(f"Service rate: {service_rate_per_hour:.1f} jobs/hour")
print(f"Current utilization: {utilization:.2%}")
print(f"Improved service time: {improved_service_time:.2f} minutes/job")
print(f"Improved utilization: {improved_utilization:.2%}")
print("Lower service time increases capacity and reduces pressure.")

```

## **3. Scenario Comparison**

### **3.1. Decision Criteria**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780943342" width="250">



>* Criteria guide disciplined scenario comparisons
>* Measures include costs, risks, and service effects

>* Balance performance, cost, risk, and feasibility
>* Avoid optimizing one measure causing bigger problems

>* Match criteria to context and priorities
>* State assumptions for transparent scenario choices



```
#@title Python Code - Decision Criteria

# Compare operational scenarios using decision criteria.
# Keep assumptions visible for engineering decisions.
# Use simple weighted scores for ranking.

# Store scenario measures in dictionaries.
scenarios = {
    "Current": {"cost": 420, "wait": 9.0, "stockout": 0.08, "feasible": 1},
    "Extra cashier": {"cost": 500, "wait": 4.5, "stockout": 0.08, "feasible": 1},
    "Higher inventory": {"cost": 470, "wait": 9.0, "stockout": 0.03, "feasible": 0},
}

# Define weights that sum to one.
weights = {"cost": 0.30, "wait": 0.35, "stockout": 0.25, "feasible": 0.10}

# Convert lower-is-better measures into scores.
def lower_score(value, best, worst):
    if worst == best:
        return 1.0
    return (worst - value) / (worst - best)

# Collect ranges for each criterion.
costs = [item["cost"] for item in scenarios.values()]
waits = [item["wait"] for item in scenarios.values()]
risks = [item["stockout"] for item in scenarios.values()]

# Score each scenario using stated assumptions.
results = {}
for name, item in scenarios.items():
    cost_score = lower_score(item["cost"], min(costs), max(costs))
    wait_score = lower_score(item["wait"], min(waits), max(waits))
    risk_score = lower_score(item["stockout"], min(risks), max(risks))
    feasible_score = item["feasible"]

    total = (
        weights["cost"] * cost_score
        + weights["wait"] * wait_score
        + weights["stockout"] * risk_score
        + weights["feasible"] * feasible_score
    )
    results[name] = round(total, 3)

# Identify the preferred scenario.
best_name = max(results, key=results.get)

# Print compact comparison output.
print("Decision criteria: cost, wait, stockout risk, feasibility")
print("Weights:", weights)
for name, score in results.items():
    print(f"{name}: weighted score = {score}")
print(f"Recommended scenario: {best_name}")

```

### **3.2. Cost Tradeoffs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780943341" width="250">



>* Balance inventory costs against stockout risks
>* Balance capacity costs against waiting risks

>* Separate measurable and hidden costs
>* State assumptions and qualitative effects

>* Test scenarios against changing assumptions
>* Compare costs, service, and operating conditions



```
#@title Python Code - Cost Tradeoffs

# Compare cost tradeoffs across operating scenarios.
# Use simple formulas and stated assumptions.
# Keep output short for classroom discussion.

# Import plotting tools for one visual comparison.
import matplotlib.pyplot as plt

# Store shared demand and cost assumptions.
daily_demand = 120
unit_holding_cost = 0.18
stockout_penalty = 45

# Store queue valuation assumptions.
waiting_cost_per_hour = 28
work_hours_per_day = 8
service_rate_per_server = 18

# Create three scenario dictionaries.
scenarios = [
    {"name": "Lean", "stock": 100, "servers": 1, "server_cost": 90},
    {"name": "Balanced", "stock": 130, "servers": 2, "server_cost": 180},
    {"name": "High service", "stock": 160, "servers": 3, "server_cost": 270},
]

# Define a simple inventory cost function.
def inventory_cost(stock, demand):
    shortage = max(demand - stock, 0)
    excess = max(stock - demand, 0)

    holding_cost = excess * unit_holding_cost
    shortage_cost = shortage * stockout_penalty
    return holding_cost + shortage_cost

# Define an approximate queue waiting cost function.
def queue_waiting_cost(servers):
    capacity = servers * service_rate_per_server * work_hours_per_day
    utilization = daily_demand / capacity

    if utilization >= 1:
        return 999
    waiting_hours = utilization / (servers * service_rate_per_server)
    return daily_demand * waiting_hours * waiting_cost_per_hour

# Evaluate all scenarios using the assumptions.
results = []
for scenario in scenarios:
    inv_cost = inventory_cost(scenario["stock"], daily_demand)
    wait_cost = queue_waiting_cost(scenario["servers"])

    total_cost = inv_cost + wait_cost + scenario["server_cost"]
    results.append((scenario["name"], total_cost, inv_cost, wait_cost))

# Select the lowest total cost scenario.
best = min(results, key=lambda row: row[1])
print("Scenario cost tradeoff summary:")

# Print a compact comparison table.
for name, total, inv_cost, wait_cost in results:
    line = f"{name}: total=${total:.2f}, inventory=${inv_cost:.2f}, waiting=${wait_cost:.2f}"
    print(line)

# Print the recommended scenario.
print(f"Recommended under these assumptions: {best[0]}")

# Build values for one compact chart.
names = [row[0] for row in results]
totals = [row[1] for row in results]

# Plot the total daily cost comparison.
plt.figure(figsize=(6, 3))
plt.bar(names, totals, color=["tomato", "steelblue", "seagreen"])
plt.ylabel("Estimated daily cost, dollars")
plt.title("Inventory and Queue Cost Tradeoffs")

# Display the chart for decision comparison.
plt.tight_layout()
plt.show()

```

### **3.3. Scenario Decision Logic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780943344" width="250">



>* Compare options using stated assumptions
>* Choose the scenario matching organizational priorities

>* Reject scenarios that fail required constraints
>* Compare feasible options using agreed criteria

>* Test how uncertainty affects scenario choices
>* Explain assumptions, tradeoffs, and evidence clearly



```
#@title Python Code - Scenario Decision Logic

# Compare queue staffing scenarios clearly.
# Check feasibility before choosing preference.
# Explain assumptions using simple calculations.

# Define shared arrival and service assumptions.
arrival_rate = 18.0
service_rate_each = 12.0
max_wait_minutes = 8.0

# Define labor cost for each desk count.
labor_costs = {1: 25, 2: 50, 3: 75}

# Estimate waiting time with a simple formula.
def estimate_wait(arrival, service, desks):
    capacity = service * desks
    utilization = arrival / capacity

    if utilization >= 1.0:
        return utilization, float("inf")

    wait_hours = utilization / (capacity - arrival)
    wait_minutes = wait_hours * 60.0
    return utilization, wait_minutes

# Compare scenarios against the service constraint.
results = []
for desks, cost in labor_costs.items():
    utilization, wait = estimate_wait(
        arrival_rate, service_rate_each, desks
    )

    feasible = wait <= max_wait_minutes
    score = cost if feasible else float("inf")
    results.append((score, desks, utilization, wait, cost, feasible))

# Choose the lowest cost feasible scenario.
best = min(results, key=lambda row: row[0])

# Print a compact decision table.
print("Scenario decision logic for clinic reception")
print("Desk  Utilization  Wait(min)  Cost/hr  Feasible")

for score, desks, util, wait, cost, feasible in results:
    wait_text = "unstable" if wait == float("inf") else f"{wait:5.1f}"
    print(f"{desks:4d}  {util:10.2f}  {wait_text:>8}  ${cost:6.0f}  {feasible}")

# Print the recommendation and its condition.
print(f"Recommend {best[1]} desks under the stated assumptions.")

```
