## **1. Ordered Industrial Data**

### **1.1. Creating Ordered Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780936797" width="250">



>* Order gives industrial data meaning
>* Lists and tuples store positioned values

>* Define what each sequence position means
>* Use order to improve engineering decisions

>* Keep sequence positions consistent and comparable
>* Use order to clarify engineering context



```
#@title Python Code - Creating Ordered Records

# Ordered records keep industrial data meaningful.
# Lists store changeable sequences of observations.
# Tuples store fixed sequences of facts.

# Create a workstation route list.
workstation_route = ["Cut", "Drill", "Inspect", "Pack"]

# Create hourly output counts.
hourly_output = [42, 45, 43, 47]

# Create a fixed part record.
part_record = ("PX-104", "Line 2", 12.5, "inches")

# Print the ordered workstation path.
print("Route order:", workstation_route)

# Print the first and last steps.
print("First step:", workstation_route[0])
print("Last step:", workstation_route[-1])

# Summarize the hourly output list.
total_output = sum(hourly_output)
average_output = total_output / len(hourly_output)

# Print compact production summaries.
print("Hourly counts:", hourly_output)
print("Total output:", total_output)
print("Average per hour:", average_output)

# Update one list value safely.
hourly_output[2] = 44

# Print the corrected ordered list.
print("Corrected counts:", hourly_output)

# Print the immutable tuple record.
print("Fixed part record:", part_record)

```

### **1.2. Recording Ordered Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780936798" width="250">



>* Order gives industrial data its meaning
>* Lists and tuples preserve sequence clearly

>* Consistent order makes records easier to compare
>* Sequence preserves process meaning and trends

>* Use lists for growing observations
>* Use tuples for stable structured records



```
#@title Python Code - Recording Ordered Data

# Lists preserve changing ordered process observations.
# Tuples preserve fixed structured engineering records.
# This example compares both sequence choices.

# Create cycle times in observed order.
cycle_times_seconds = [42.5, 41.8, 43.2, 42.1]

# Create a fixed part record tuple.
part_record = ("P-104", 12.7, 6.4, 2.5)

# Access positions using zero based indexing.
first_cycle = cycle_times_seconds[0]
last_cycle = cycle_times_seconds[-1]

# Update list when another observation arrives.
cycle_times_seconds.append(42.9)
average_cycle = sum(cycle_times_seconds) / len(cycle_times_seconds)

# Summarize ordered observations with built-ins.
fastest_cycle = min(cycle_times_seconds)
slowest_cycle = max(cycle_times_seconds)

# Print compact results for learners.
print("Ordered cycle times:", cycle_times_seconds)
print("First and last cycle:", first_cycle, last_cycle)
print("Average cycle time:", round(average_cycle, 2), "seconds")
print("Fastest and slowest:", fastest_cycle, slowest_cycle)

# Show tuple positions as fixed meanings.
print("Part ID from tuple:", part_record[0])
print("Length, width, height cm:", part_record[1:])

```

### **1.3. Industrial Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780936800" width="250">



>* Sequences keep industrial data in meaningful order
>* Order supports analysis and decisions

>* Sequences organize repeated observations and procedure steps
>* Consistent positions prevent misleading data interpretation

>* Use lists for growing industrial observations
>* Use tuples for fixed ordered records



```
#@title Python Code - Industrial Sequences

# Industrial sequences preserve meaningful order.
# Lists can grow during operations.
# Tuples can store fixed records.

# Create ordered cycle times as a list.
cycle_times_seconds = [52, 55, 53, 58, 54]

# Create fixed part dimensions as tuple.
part_dimensions_mm = (120.0, 45.0, 8.0)

# Access positions with zero-based indexing.
first_cycle = cycle_times_seconds[0]
last_cycle = cycle_times_seconds[-1]

# Update list data after review.
cycle_times_seconds[3] = 56

# Summarize the ordered list values.
average_cycle = sum(cycle_times_seconds) / len(cycle_times_seconds)

# Label tuple positions consistently.
length_mm, width_mm, thickness_mm = part_dimensions_mm

# Show compact engineering sequence results.
print("Cycle times list:", cycle_times_seconds)
print("First and last cycle:", first_cycle, last_cycle)
print("Average cycle time:", round(average_cycle, 1), "seconds")
print("Part dimensions tuple:", part_dimensions_mm)
print("Length, width, thickness:", length_mm, width_mm, thickness_mm)

```

## **2. Sequence Operations**

### **2.1. Sequence Summary**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780936791" width="250">



>* Sequences keep industrial data in meaningful order
>* Ordered values support access, updates, and summaries

>* Access and edit ordered data by position
>* Use functions to count, total, compare

>* Use summaries to support engineering decisions
>* Check context, quality, and possible outliers



```
#@title Python Code - Sequence Summary

# Summarize ordered production data safely.
# Built-in functions reveal quick sequence insights.
# Lists update, while tuples preserve records.

# Store hourly output as a list.
hourly_output = [42, 45, 39, 48, 46, 44]

# Correct one validated data entry.
hourly_output[2] = 41

# Store fixed inspection limits as a tuple.
weight_limits_grams = (95, 105)

# Summarize the updated sequence.
reading_count = len(hourly_output)
total_units = sum(hourly_output)
lowest_hour = min(hourly_output)
highest_hour = max(hourly_output)

# Calculate average with safe division.
average_units = total_units / reading_count

# Print concise engineering summaries.
print("Hourly output list:", hourly_output)
print("Fixed weight limits, grams:", weight_limits_grams)
print("Number of hourly readings:", reading_count)
print("Total units produced:", total_units)
print("Lowest hourly output:", lowest_hour)
print("Highest hourly output:", highest_hour)
print("Average hourly output:", round(average_units, 1))

```

### **2.2. Accessing Sequence Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780936793" width="250">



>* Use indexes to retrieve ordered sequence values
>* Zero-based positions show steps from the start

>* Use slicing to study ordered data ranges
>* Use negative indexes for recent values

>* Match each index to its meaning
>* Check length before retrieving sequence values



```
#@title Python Code - Accessing Sequence Data

# Sequence positions support engineering data access.
# Lists and tuples store ordered records.
# Indexing retrieves values by exact position.

# Store hourly machine temperatures in Celsius.
temperatures_c = [68.5, 69.1, 70.0, 71.3, 70.8, 69.9]

# Store fixed station names as a tuple.
stations = ("Cut", "Drill", "Inspect", "Pack")

# Check sequence sizes before indexing.
valid_temperature_data = len(temperatures_c) >= 3

# Access values using zero-based positions.
first_temperature = temperatures_c[0]
third_temperature = temperatures_c[2]
latest_temperature = temperatures_c[-1]

# Slice a useful start-up window.
startup_temperatures = temperatures_c[:3]

# Summarize sequence data with built-ins.
average_startup = sum(startup_temperatures) / len(startup_temperatures)
highest_temperature = max(temperatures_c)
lowest_temperature = min(temperatures_c)

# Access tuple elements without changing them.
first_station = stations[0]
last_station = stations[-1]

# Print concise teaching results.
print("Enough temperature readings:", valid_temperature_data)
print("First, third, latest temperatures:", first_temperature, third_temperature, latest_temperature)
print("Startup temperature slice:", startup_temperatures)
print("Average startup temperature:", round(average_startup, 2))
print("Lowest and highest temperatures:", lowest_temperature, highest_temperature)
print("First and last stations:", first_station, last_station)

```

### **2.3. Sequence Summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780936795" width="250">



>* Summaries turn ordered data into decisions
>* Built-in functions reveal counts, extremes, totals

>* Interpret summaries using data context
>* Match summary functions to data type

>* Verify sequence context before summarizing
>* Use summaries to communicate operational insights



```
#@title Python Code - Sequence Summaries

# Summaries convert sequence values into decisions.
# Industrial engineers summarize ordered observations often.
# Built-in functions provide quick operational insight.

# Create ordered cycle time observations.
cycle_times_seconds = [52, 49, 55, 51, 53]

# Check how many observations were collected.
observation_count = len(cycle_times_seconds)

# Find fastest and slowest observed times.
fastest_time = min(cycle_times_seconds)
slowest_time = max(cycle_times_seconds)

# Add all observed processing time.
total_time = sum(cycle_times_seconds)

# Compute a typical cycle time.
average_time = total_time / observation_count

# Show concise summaries for operations.
print("Cycle time observations:", observation_count)
print("Fastest time in seconds:", fastest_time)
print("Slowest time in seconds:", slowest_time)
print("Total time in seconds:", total_time)
print("Average time in seconds:", round(average_time, 1))

# Summarize an immutable approved standard.
standard_batch_sizes = (100, 100, 120, 110)
standard_total = sum(standard_batch_sizes)

# Compare list and tuple summaries.
print("Approved batch entries:", len(standard_batch_sizes))
print("Approved total units:", standard_total)

```

## **3. Engineering Sequence Choices**

### **3.1. Mutable Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780936787" width="250">



>* Use lists for changing engineering sequences
>* Keep operational records aligned with current conditions

>* Mutable lists support changing daily work records
>* Shared updates need ownership and documentation

>* Use mutable records for changing operations
>* Preserve logs for important final states



```
#@title Python Code - Mutable Records

# Lists model records that change.
# Industrial queues often need updates.
# Mutability supports live operational decisions.

# Create a mutable inspection schedule list.
inspection_queue = ["Pump-1", "Conveyor-2", "Mixer-3"]

# Show the starting working record.
print("Start:", inspection_queue)

# Insert an urgent machine earlier.
inspection_queue.insert(1, "Press-9")

# Replace a machine after reassignment.
inspection_queue[0] = "Pump-1A"

# Remove a completed inspection task.
completed_task = inspection_queue.pop(2)

# Add a newly reported machine.
inspection_queue.append("Robot-4")

# Summarize the current live record.
print("Completed:", completed_task)
print("Now:", inspection_queue)
print("Next task:", inspection_queue[0])
print("Tasks left:", len(inspection_queue))

# Compare with an immutable approval tuple.
approved_route = ("Dock A", "Aisle 3", "Pack 2")

# Tuples communicate stable historical records.
print("Approved route:", approved_route)

```

### **3.2. Immutable Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780936785" width="250">



>* Tuples store fixed engineering facts safely
>* Stable values preserve accuracy and traceability

>* Tuples signal stable, unedited records
>* Immutability preserves raw data integrity

>* Keep shared record identities stable
>* Create new sequences for revisions



```
#@title Python Code - Immutable Records

# Tuples protect stable engineering records.
# Lists support records needing edits.
# This example compares both choices.

# Create an immutable machine location record.
machine_location_ft = (12.5, 8.0, 0.0)

# Create a mutable daily production count list.
daily_counts = [118, 121, 117, 124]

# Show the original stable location record.
print("Machine location tuple:", machine_location_ft)

# Show the editable production list.
print("Daily counts list:", daily_counts)

# Update the list because production changed.
daily_counts[2] = 120

# Summarize the updated mutable sequence.
average_count = sum(daily_counts) / len(daily_counts)

# Create a revised tuple instead of editing original.
revised_location_ft = (12.5, 8.5, 0.0)

# Compare original and revised immutable records.
print("Updated counts list:", daily_counts)
print("Average count:", round(average_count, 1))
print("Original location kept:", machine_location_ft)
print("Revised location tuple:", revised_location_ft)

# Demonstrate tuple immutability safely with try.
try:
    machine_location_ft[1] = 8.5
except TypeError as error:
    print("Tuple edit blocked:", type(error).__name__)

# Store identification fields as a reliable tuple.
batch_key = ("Line-2", "Shift-B", "Part-104")

# Print the stable grouped identifier.
print("Stable batch key:", batch_key)

```

### **3.3. Stable Sequence Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780936789" width="250">



>* Fixed order protects completed engineering records
>* Stable sequences support audits and traceability

>* Use tuples for finalized grouped records
>* Preserve field order across data pipelines

>* Separate changing operations from protected archives
>* Revise records transparently instead of silently altering



```
#@title Python Code - Stable Sequence Records

# Stable records protect finalized engineering data.
# Tuples keep field order and values fixed.
# Lists remain useful for active updates.

# Create a finalized inspection tuple.
inspection_record = ("P-104", "Station-2", 2.503, "pass")

# Create an active job list.
active_jobs = ["cut", "drill", "inspect", "pack"]

# Update the active planning list.
active_jobs[1] = "ream"

# Summarize the stable tuple fields.
part_id, station_id, diameter_in, result = inspection_record

# Try changing the stable record safely.
try:
    inspection_record[2] = 2.499

except TypeError as error:
    tuple_message = "Tuple blocked accidental change."

# Create a revised tuple deliberately.
revised_record = (part_id, station_id, 2.499, result)

# Print concise teaching results.
print("Stable tuple:", inspection_record)
print("Updated list:", active_jobs)
print("Diameter inches:", diameter_in)
print(tuple_message)
print("Revised tuple:", revised_record)
print("Tuple length:", len(inspection_record))

```
