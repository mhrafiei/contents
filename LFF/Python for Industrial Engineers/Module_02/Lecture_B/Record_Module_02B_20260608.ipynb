## **1. Key Value Maps**

### **1.1. Machine Identifier Maps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780938365" width="250">



>* Machine IDs link to specific machine details
>* Dictionaries retrieve asset information directly

>* Map each machine ID to consistent details
>* Retrieve equipment information for maintenance and reporting

>* Unique keys prevent machine record confusion
>* Consistent identifiers support reliable equipment data



```
#@title Python Code - Machine Identifier Maps

# Dictionaries connect identifiers to machine information.
# Keys should uniquely name physical equipment.
# Values store useful engineering status details.

# Create a small machine identifier map.
machine_status = {
    "LN1-PRESS-A": "active",
    "LN1-PRESS-B": "idle",

    "LN2-ROBOT-1": "under maintenance",
    "UTIL-PUMP-3": "out of service",
}

# Retrieve one machine status using its identifier.
selected_id = "LN2-ROBOT-1"
selected_status = machine_status[selected_id]

# Add another machine identifier and status.
machine_status["QC-CAMERA-2"] = "active"

# Update a status after a maintenance event.
machine_status["UTIL-PUMP-3"] = "under maintenance"

# Count the unique status labels with a set.
unique_statuses = set(machine_status.values())

# Print compact results for the learner.
print("Selected machine:", selected_id)
print("Current status:", selected_status)
print("Total machines tracked:", len(machine_status))
print("Unique statuses:", sorted(unique_statuses))

# Show direct lookup for every machine.
for machine_id, status in machine_status.items():
    print(machine_id, "->", status)

```

### **1.2. Part Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780938367" width="250">



>* Part IDs link to useful engineering values
>* Dictionaries provide fast access to part data

>* Match part values to the task
>* Keep dictionaries focused and easy to interpret

>* Use consistent part identifiers and value formats
>* Reliable dictionaries prevent costly engineering mistakes



```
#@title Python Code - Part Values

# Dictionaries connect identifiers to engineering values.
# Part numbers can become reliable keys.
# Values answer one focused operational question.

# Create a dictionary mapping parts to unit costs.
part_unit_cost_usd = {
    "BRK-101": 12.75,
    "SHAFT-204": 48.20,

    "PLATE-330": 31.10,
    "SEAL-415": 6.40,
}

# Look up one known part directly.
selected_part = "SHAFT-204"
selected_cost = part_unit_cost_usd[selected_part]

# Use get for safer unknown part lookups.
missing_part = "BOLT-999"
fallback_message = "part not found"
missing_result = part_unit_cost_usd.get(missing_part, fallback_message)

# Add a new part value consistently.
part_unit_cost_usd["PIN-502"] = 3.85
part_count = len(part_unit_cost_usd)

# Compute a simple purchasing total.
order_quantities = {"BRK-101": 8, "PIN-502": 40}
order_total = sum(part_unit_cost_usd[p] * q for p, q in order_quantities.items())

# Print compact results for the learner.
print(f"Cost for {selected_part}: ${selected_cost:.2f}")
print(f"Lookup for {missing_part}: {missing_result}")
print(f"Tracked part identifiers: {part_count}")
print(f"Small order total: ${order_total:.2f}")

```

### **1.3. Dictionary Lookups**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780938368" width="250">



>* Use identifiers to retrieve connected values
>* Keys help find records quickly

>* Consistent identifiers make lookups reliable
>* Mistyped keys cause missing or wrong results

>* Identifiers connect to specific engineering information
>* Lookups turn keys into useful decisions



```
#@title Python Code - Dictionary Lookups

# Dictionaries connect identifiers with operational values.
# Lookups retrieve values using exact keys.
# Consistent engineering identifiers prevent lookup failures.

# Build a small machine status dictionary.
machine_status = {
    "Press Line 3": "scheduled maintenance",
    "Robot Arm A2": "running normally",

    "Pump P-104": "needs vibration check",
}

# Look up values using known identifiers.
press_status = machine_status["Press Line 3"]
pump_status = machine_status["Pump P-104"]

# Use get for safer missing-key lookups.
missing_status = machine_status.get(
    "Motor-17",
    "identifier not found",
)

# Print concise lookup results.
print("Press Line 3:", press_status)
print("Pump P-104:", pump_status)
print("Motor-17:", missing_status)

# Add capacities using metric production units.
hourly_capacity = {
    "Press Line 3": 120,
    "Robot Arm A2": 85,

    "Pump P-104": 40,
}

# Retrieve capacity for a planning calculation.
selected_machine = "Robot Arm A2"
selected_capacity = hourly_capacity[selected_machine]

# Print the selected machine capacity.
print(selected_machine, "capacity:", selected_capacity, "units per hour")

```

## **2. Unique Resource Sets**

### **2.1. Set Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780938372" width="250">



>* Sets keep only distinct resource identifiers
>* Unique items make engineering summaries clearer

>* Sets track membership, not order or duplicates
>* Use sets to compare unique resources

>* Sets turn noisy records into unique inventories
>* Unique resources support comparison and decisions



```
#@title Python Code - Set Basics

# Sets store unique engineering resource identifiers.
# Repeated observations become one membership list.
# Membership checks answer simple resource questions.

# Create repeated workstation observations.
station_scans = ["CUT-1", "WELD-2", "CUT-1", "PACK-4", "WELD-2"]

# Convert observations into unique resource identifiers.
unique_stations = set(station_scans)

# Add an already-seen station identifier.
unique_stations.add("CUT-1")

# Add a new inspection station identifier.
unique_stations.add("INSP-3")

# Check whether one resource appeared.
welder_seen = "WELD-2" in unique_stations

# Sort only for readable classroom output.
sorted_stations = sorted(unique_stations)

# Print the raw observations first.
print("Raw scan count:", len(station_scans))

# Print the unique resource count.
print("Unique station count:", len(unique_stations))

# Print the unique identifiers clearly.
print("Unique stations:", sorted_stations)

# Print one membership result.
print("Was WELD-2 observed?", welder_seen)

```

### **2.2. Set Operations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780938370" width="250">



>* Sets track resource membership, not order
>* Operations reveal availability, overlap, and gaps

>* Combine, compare, and separate resource sets
>* Reveal shared items, shortages, and availability

>* Compare unique resources across different records
>* Find mismatches for clearer operational decisions



```
#@title Python Code - Set Operations

# Sets answer unique resource questions.
# Operations compare required and available resources.
# This example uses maintenance tool sets.

# Define resources required by one maintenance task.
required_tools = {"torque wrench", "caliper", "safety glasses", "lift cart"}

# Define tools currently available in storeroom.
available_tools = {"caliper", "safety glasses", "socket set", "lift cart"}

# Combine both collections into unique tools.
all_tools = required_tools | available_tools

# Find tools appearing in both sets.
covered_tools = required_tools & available_tools

# Find required tools missing from storeroom.
missing_tools = required_tools - available_tools

# Find tools present in only one set.
one_side_only = required_tools ^ available_tools

# Print concise results for engineering review.
print("Required tools:", sorted(required_tools))
print("Available tools:", sorted(available_tools))
print("All unique tools:", sorted(all_tools))
print("Covered required tools:", sorted(covered_tools))
print("Missing required tools:", sorted(missing_tools))
print("Only in one collection:", sorted(one_side_only))

```

### **2.3. Resource Set Comparisons**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780938374" width="250">



>* Sets compare unique resources across collections
>* Overlaps reveal urgent maintenance needs

>* Compare sets to find gaps and matches
>* Use results for practical resource decisions

>* Compare records to find compliance gaps
>* Use overlaps and differences for coordination



```
#@title Python Code - Resource Set Comparisons

# Compare resource collections using Python sets.
# Sets keep each resource identifier once.
# Differences reveal missing and extra resources.

# Define required tools for one assembly cell.
required_tools = {"torque_wrench", "fixture_A", "caliper", "barcode_scanner"}

# Define tools currently found at that workstation.
available_tools = {"fixture_A", "caliper", "caliper", "spare_bit"}

# Find tools that are both required and available.
ready_tools = required_tools & available_tools

# Find required tools missing from the station.
missing_tools = required_tools - available_tools

# Find unexpected tools currently at the station.
extra_tools = available_tools - required_tools

# Check whether the workstation is fully prepared.
is_ready = missing_tools == set()

print("Required unique tools:", sorted(required_tools))
print("Available unique tools:", sorted(available_tools))
print("Ready tools:", sorted(ready_tools))
print("Missing tools:", sorted(missing_tools))
print("Extra tools:", sorted(extra_tools))
print("Workstation ready:", is_ready)

```

## **3. Combined Data Records**

### **3.1. Nested Data Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780938376" width="250">



>* Nested records organize complex engineering information
>* Inner details support planning and decisions

>* Nested records organize differently shaped data
>* Related details stay grouped with their item

>* Keep nested records clear and consistent
>* Group related details for useful analysis



```
#@title Python Code - Nested Data Records

# Nested records organize related engineering details.
# Dictionaries can contain lists and sets.
# Small records support operational reasoning.

# Create one nested equipment record.
press_record = {
    "asset_id": "PRESS-014",
    "location": "Line A"
}

# Add nested operating and supplier details.
press_record["operating"] = {
    "status": "running",
    "cycle_time_seconds": 42
}

press_record["supplier"] = {
    "company": "Metro Hydraulics",
    "region": "Midwest"
}

# Store repeated readings in a list.
press_record["temperature_c"] = [64.2, 65.1, 66.0]

# Store approved operators as unique names.
press_record["approved_operators"] = {"Amina", "Luis", "Chen"}

# Build a second record for comparison.
conveyor_record = {
    "asset_id": "CONV-008",
    "location": "Line B"
}

conveyor_record["approved_operators"] = {"Luis", "Maya", "Chen"}

# Compare operator sets between resources.
shared_operators = (
    press_record["approved_operators"]
    & conveyor_record["approved_operators"]
)

# Read values from different nesting levels.
asset_code = press_record["asset_id"]
status_text = press_record["operating"]["status"]

# Compute one summary from nested readings.
average_temperature = (
    sum(press_record["temperature_c"])
    / len(press_record["temperature_c"])
)

# Print a compact record summary.
print("Asset:", asset_code)
print("Status:", status_text)
print("Supplier:", press_record["supplier"]["company"])
print("Average temperature C:", round(average_temperature, 1))
print("Shared operators:", sorted(shared_operators))

```

### **3.2. Linked Data Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780938378" width="250">



>* Store related industrial records separately
>* Connect records using shared identifiers

>* Reduce repeated details using shared identifiers
>* Link structures to model real relationships

>* Use consistent identifiers for reliable record links
>* Separate record details from relationship links



```
#@title Python Code - Linked Data Records

# Linked records connect industrial information clearly.
# Shared identifiers reduce repeated engineering details.
# Sets help compare related resource collections.

# Create equipment records keyed by stable identifiers.
equipment = {
    "PUMP-01": {"area": "Mixing", "power_hp": 15},
    "MOTOR-17": {"area": "Packing", "power_hp": 5},
    "SENS-03": {"area": "Drying", "range_c": 120},
}

# Create work orders that reference equipment identifiers.
work_orders = {
    "WO-1001": {"asset": "PUMP-01", "tech": "Ava"},
    "WO-1002": {"asset": "MOTOR-17", "tech": "Noah"},
    "WO-1003": {"asset": "PUMP-01", "tech": "Ava"},
}

# Use sets for unique linked asset identifiers.
registered_assets = set(equipment.keys())
assigned_assets = {record["asset"] for record in work_orders.values()}
missing_records = assigned_assets - registered_assets

# Count work orders per linked asset.
asset_order_counts = {}
for record in work_orders.values():
    asset_code = record["asset"]
    asset_order_counts[asset_code] = asset_order_counts.get(asset_code, 0) + 1

# Print compact results from linked records.
print("Registered asset identifiers:", sorted(registered_assets))
print("Assigned asset identifiers:", sorted(assigned_assets))
print("Missing equipment records:", sorted(missing_records))

# Display summarized linked work order counts.
for asset_code, order_count in sorted(asset_order_counts.items()):
    asset_area = equipment[asset_code]["area"]
    print(asset_code, "in", asset_area, "has", order_count, "work orders")

```

### **3.3. Structured Dataset Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780938380" width="250">



>* Group related engineering details consistently
>* Make datasets readable, updateable, and shareable

>* Use consistent record patterns for every entry
>* Combine dictionaries, lists, and sets meaningfully

>* Structured records guide engineering decisions
>* Choose dictionaries, lists, and sets deliberately



```
#@title Python Code - Structured Dataset Records

# Structured records combine several built-in structures.
# Dictionaries name fields for each engineering record.
# Sets keep repeated categories uniquely represented.

# Build a tiny maintenance dataset.
records = [
    {
        "asset_id": "P-101",
        "area": "North pump room",
        "status": "running",

        "last_inspection": "2026-05-20",
        "technician": "Amina",
        "fault_flags": {"seal leak", "vibration"},
        "parts_needed": ["gasket", "bearing"],
    },

    {
        "asset_id": "P-102",
        "area": "North pump room",
        "status": "stopped",
        "last_inspection": "2026-05-12",

        "technician": "Diego",
        "fault_flags": {"vibration", "missing guard"},
        "parts_needed": ["guard", "bearing"],
    },

    {
        "asset_id": "C-201",
        "area": "Compressor bay",
        "status": "running",
        "last_inspection": "2026-05-18",

        "technician": "Amina",
        "fault_flags": set(),
        "parts_needed": ["filter"],
    },
]

# Confirm each record has the required fields.
required_fields = {
    "asset_id", "area", "status", "last_inspection",
    "technician", "fault_flags", "parts_needed"
}

# Keep only records with complete expected structure.
valid_records = [
    record for record in records
    if required_fields.issubset(record.keys())
]

# Combine sets across records for unique issues.
all_fault_flags = set().union(
    *(record["fault_flags"] for record in valid_records)
)

# Count unique parts by converting lists into sets.
unique_parts = set().union(
    *(set(record["parts_needed"]) for record in valid_records)
)

# Find stopped assets using dictionary fields.
stopped_assets = [
    record["asset_id"] for record in valid_records
    if record["status"] == "stopped"
]

# Print a compact engineering summary.
print("Valid structured records:", len(valid_records))
print("Unique fault flags:", sorted(all_fault_flags))
print("Unique required parts:", sorted(unique_parts))
print("Stopped assets:", stopped_assets)
print("First record technician:", valid_records[0]["technician"])

```
