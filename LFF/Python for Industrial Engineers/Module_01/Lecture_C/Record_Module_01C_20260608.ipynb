## **1. Conditional Rules**

### **1.1. Basic Condition Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780935769" width="250">



>* Check conditions against engineering limits
>* Use true or false to choose actions

>* Write clear, testable comparison rules
>* Match conditions to real classification boundaries

>* Check limits and boundary cases carefully
>* Validate data before applying engineering rules



```
#@title Python Code - Basic Condition Checks

# Basic checks translate engineering rules.
# Comparisons produce true or false.
# Boundaries need careful interpretation.

# Define one measured process temperature.
temperature_c = 82.0
# Define the maximum permitted temperature.
maximum_c = 85.0

# Check whether the temperature is acceptable.
if temperature_c <= maximum_c:
    status = "acceptable"
else:
    status = "review needed"

# Report the classification result.
print("Temperature status:", status)
# Define one measured pipe pressure.
pressure_psi = 128.0
# Define the low warning pressure.
low_limit_psi = 120.0

# Define the high warning pressure.
high_limit_psi = 150.0
# Check whether pressure stays inside limits.
if low_limit_psi <= pressure_psi <= high_limit_psi:
    pressure_status = "inside range"
else:
    pressure_status = "outside range"

# Report the pressure classification.
print("Pressure status:", pressure_status)
# Define a sensor reading state.
sensor_online = True
# Check whether analysis may continue.
if sensor_online:
    sensor_message = "continue monitoring"
else:
    sensor_message = "pause analysis"

# Report the sensor decision.
print("Sensor decision:", sensor_message)
# Show the exact boundary behavior.
boundary_c = 85.0
# Compare equality against the limit.
if boundary_c <= maximum_c:
    boundary_status = "still permitted"
else:
    boundary_status = "too high"

# Print the boundary classification.
print("Boundary result:", boundary_status)
# Finish with a traceable summary.
print("Checked temperature, pressure, sensor, and boundary rules.")

```

### **1.2. Condition Chains**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1780935767" width="250">



>* Test categories in a planned sequence
>* Use the first matching rule

>* Test specific or urgent conditions first
>* Set clear thresholds for engineering categories

>* Show decision assumptions in each branch
>* Trace sample values to catch logic errors



```
#@title Python Code - Condition Chains

# Classify process readings using ordered condition chains.
# Specific conditions should appear before broad conditions.
# Trace sample values to verify boundary behavior.

# Store tank levels measured in feet.
tank_levels_ft = [1.5, 3.0, 8.0, 12.0, 16.5]

# Define a function with chained conditions.
def classify_tank_level(level_ft):
    # Check invalid sensor values first.
    if level_ft < 0:
        return "sensor fault"

    # Check urgent low level next.
    elif level_ft < 2:
        return "shutdown: too low"

    # Check warning range before normal range.
    elif level_ft < 5:
        return "warning: low level"

    # Check broad normal operating range.
    elif level_ft <= 14:
        return "normal operation"

    # Handle remaining high readings last.
    else:
        return "shutdown: too high"

# Print a compact trace table.
print("Tank level condition chain trace")

# Trace each measurement through the chain.
for level_ft in tank_levels_ft:
    label = classify_tank_level(level_ft)

    # Show input value and selected category.
    print(f"{level_ft:4.1f} ft -> {label}")

# Compare one boundary value explicitly.
boundary_label = classify_tank_level(14.0)

# Print the boundary classification decision.
print(f"Boundary 14.0 ft -> {boundary_label}")

```

### **1.3. Branching Decisions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1780935771" width="250">



>* Branches choose paths using tested conditions
>* They classify engineering states and actions

>* Test critical categories before broad ones
>* Check for missing or misordered cases

>* Handle threshold values deliberately
>* Trace branches for reliable classifications



```
#@title Python Code - Branching Decisions

# Branches classify process readings into actions.
# Threshold order controls the final decision.
# Boundary values deserve explicit engineering attention.

# Store example temperatures in degrees Celsius.
temperatures_c = [68.0, 75.0, 85.0, 92.0]

# Store process labels for readable output.
labels = ["Reactor A", "Reactor B", "Reactor C", "Reactor D"]

# Define a function for one classification rule.
def classify_temperature(temp_c):
    # Test the most critical condition first.
    if temp_c > 90.0:
        return "shutdown required"

    # Test the warning band after danger.
    elif temp_c >= 80.0:
        return "warning: elevated"

    # Treat the exact limit as still acceptable.
    elif temp_c <= 75.0:
        return "normal operation"

    # Catch remaining readings between limits.
    else:
        return "monitor closely"

# Print a compact heading for the table.
print("Temperature classification results:")

# Classify each reading using the same branch order.
for label, temp_c in zip(labels, temperatures_c):
    status = classify_temperature(temp_c)
    print(label, temp_c, "C ->", status)

# Trace one boundary case explicitly.
boundary_temp = 75.0
boundary_status = classify_temperature(boundary_temp)
print("Boundary check:", boundary_temp, "C ->", boundary_status)

```

## **2. Loop Patterns**

### **2.1. For Loop Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780935761" width="250">



>* Repeat calculations across known data items
>* Process each item clearly and consistently

>* Repeat steps for each data item
>* Loop variable tracks the current measurement

>* Indentation controls repeated loop actions
>* Place final summaries after the loop



```
#@title Python Code - For Loop Basics

# For loops repeat work across readings.
# Each item becomes the loop variable.
# Indentation controls what repeats each time.

# Store small process temperature readings.
temperatures_c = [68.5, 72.0, 79.5, 83.0, 76.5]

# Define the acceptable operating limit.
upper_limit_c = 80.0

# Count readings that exceed the limit.
high_count = 0

# Classify each reading one at a time.
for reading_c in temperatures_c:
    if reading_c <= upper_limit_c:
        status = "acceptable"
    else:
        status = "too high"

    # Update the count only when needed.
    if status == "too high":
        high_count = high_count + 1

    # Print the repeated classification result.
    print(f"{reading_c} C is {status}.")

# Print a final summary after looping.
print(f"High readings found: {high_count} of {len(temperatures_c)}.")

```

### **2.2. While Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780935763" width="250">



>* Repeat actions while a condition stays true
>* Model processes that stop at practical targets

>* Design conditions that move loops toward stopping
>* Use updates to model changing processes

>* Trace while loops to catch logic errors
>* Check boundaries for correct stopping behavior



```
#@title Python Code - While Loops

# Model a cooling tank process.
# While loops repeat until finished.
# Trace each update carefully.

# Set starting process values.
temperature_c = float(86.0)
target_c = float(60.0)
cooling_rate_c = float(7.5)
minutes_per_step = int(5)

# Track elapsed time safely.
elapsed_minutes = int(0)
step_count = int(0)
maximum_steps = int(10)

# Print the initial condition.
print("Cooling tank toward safe range")
print(f"Start: {temperature_c:.1f} C")

# Continue while temperature remains high.
while temperature_c > target_c and step_count < maximum_steps:
    step_count = step_count + 1
    elapsed_minutes = elapsed_minutes + minutes_per_step

    # Update the simulated process state.
    temperature_c = temperature_c - cooling_rate_c
    print(f"Step {step_count}: {temperature_c:.1f} C")

# Classify why the loop stopped.
if temperature_c <= target_c:
    stop_reason = "target reached"
else:
    stop_reason = "safety limit reached"

# Summarize the loop outcome.
print(f"Stopped after {elapsed_minutes} minutes")
print(f"Reason: {stop_reason}")

```

### **2.3. Accumulating Process Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780935765" width="250">



>* Loops combine measurements into useful summaries
>* Accumulators update running results consistently

>* Start accumulators with sensible initial values
>* Update them to summarize process data

>* Avoid accumulator resets, misplaced updates, poor starts
>* Trace loops to trust process summaries



```
#@title Python Code - Accumulating Process Values

# Loops can build useful process summaries.
# Accumulators remember results from earlier measurements.
# This example totals several pipe flows.

# Store small flow readings in gallons per minute.
flow_rates_gpm = [18.5, 21.0, 19.5, 22.0]

# Start the accumulator at a neutral value.
total_flow_gpm = 0.0

# Add each reading into the running total.
for flow_gpm in flow_rates_gpm:
    total_flow_gpm = total_flow_gpm + flow_gpm

# Count readings safely before averaging.
reading_count = len(flow_rates_gpm)

# Compute the average after accumulation ends.
average_flow_gpm = total_flow_gpm / reading_count

# Count readings above a practical threshold.
high_flow_count = 0

# Update the counter only when needed.
for flow_gpm in flow_rates_gpm:
    if flow_gpm > 20.0:
        high_flow_count = high_flow_count + 1

# Display concise engineering summaries.
print("Flow readings:", flow_rates_gpm)
print("Total flow:", round(total_flow_gpm, 1), "gpm")
print("Average flow:", round(average_flow_gpm, 1), "gpm")
print("Readings above 20.0 gpm:", high_flow_count)
print("Accumulator meaning: total flow entering the tank.")

```

## **3. Debugging Logic**

### **3.1. Trace Program Flow**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780935756" width="250">



>* Follow each executed step and stored value
>* Find where logic differs from intent

>* Trace small data against expected results
>* Verify working code for hidden logic errors

>* Track actual values through each step
>* Explain flow aloud to find logic gaps



```
#@title Python Code - Trace Program Flow

# Trace program flow through pressure classification.
# Each loop pass updates one counter.
# Printed trace reveals a logical error.

# Store small pressure readings in psi.
pressures_psi = [48, 52, 75, 91]

# Define intended acceptable pressure limits.
low_limit = 50
high_limit = 90

# Start attention count before tracing.
attention_count = 0
print("Trace of flawed logic:")

# Follow each reading through conditional branches.
for pressure in pressures_psi:
    print("checking", pressure, "psi")

    # This condition is flawed and too broad.
    if pressure > low_limit:
        status = "acceptable"

    elif pressure > high_limit:
        status = "high"

    else:
        status = "low"

    # Count readings requiring engineering attention.
    if status != "acceptable":
        attention_count = attention_count + 1

    print("branch result:", status)

# Show the misleading result from flawed flow.
print("flawed attention count:", attention_count)
print("91 psi was misclassified before high was checked.")

# Reset count before corrected trace.
attention_count = 0
print("Corrected logic trace:")

# Check high pressure before general acceptable case.
for pressure in pressures_psi:
    if pressure > high_limit:
        status = "high"

    elif pressure < low_limit:
        status = "low"

    else:
        status = "acceptable"

    if status != "acceptable":
        attention_count = attention_count + 1

# Show the corrected count after tracing.
print("corrected attention count:", attention_count)

```

### **3.2. Trace Branch Outcomes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780935757" width="250">



>* Follow which branch the program chooses
>* Use tracing to spot wrong decisions

>* Test boundary values around each decision.
>* Compare branch results with real rules.

>* Check condition order carefully
>* Overlapping branches can hide serious cases



```
#@title Python Code - Trace Branch Outcomes

# Trace branches using boundary process readings.
# Compare intended and actual decision paths.
# Keep output short for classroom debugging.

# Store small readings near safety thresholds.
readings_celsius = [79, 80, 90, 95, 96]

# Define a flawed classifier with overlapping conditions.
def flawed_status(temp_celsius):
    if temp_celsius >= 80:
        return "warning branch"

    elif temp_celsius >= 95:
        return "critical branch"

    else:
        return "normal branch"

# Define the corrected classifier using strict ordering.
def corrected_status(temp_celsius):
    if temp_celsius >= 95:
        return "critical branch"

    elif temp_celsius >= 80:
        return "warning branch"

    else:
        return "normal branch"

# Print a compact trace table header.
print("Temp C | flawed branch | corrected branch")

# Trace each reading through both branch structures.
for temp_celsius in readings_celsius:
    flawed_branch = flawed_status(temp_celsius)

    corrected_branch = corrected_status(temp_celsius)

    message = f"{temp_celsius:6} | {flawed_branch:13} | {corrected_branch}"

    print(message)

# Show the debugging lesson in one sentence.
print("The broad warning test hides the critical branch.")

```

### **3.3. Fixing Logic Errors**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780935759" width="250">



>* Wrong results can hide logic errors
>* Compare intended behavior with each step

>* Use evidence to locate logic mismatches
>* Make small fixes, then retest edge cases

>* Describe decisions before changing program code
>* Test small cases for reliable logic



```
#@title Python Code - Fixing Logic Errors

# Debugging logic means comparing expected behavior.
# Small engineering examples make mistakes visible.
# Boundary tests reveal incorrect program reasoning.

# Define tiny flow readings in liters per minute.
flow_readings = [79.9, 80.0, 100.0, 120.0, 120.1]

# State the intended rule before coding.
expected_labels = ["low", "normal", "normal", "normal", "high"]

# This function contains a boundary logic error.
def buggy_label(flow_rate):
    # The upper boundary is handled incorrectly.
    if flow_rate >= 120.0:
        return "high"

    # The lower boundary is handled correctly.
    if flow_rate < 80.0:
        return "low"

    # Remaining readings are considered normal.
    return "normal"

# This corrected function matches the written rule.
def fixed_label(flow_rate):
    # Values above the upper limit are high.
    if flow_rate > 120.0:
        return "high"

    # Values below the lower limit are low.
    if flow_rate < 80.0:
        return "low"

    # Boundary values stay inside normal.
    return "normal"

# Compare actual results against expected results.
buggy_labels = [buggy_label(value) for value in flow_readings]

# Identify exactly where program flow differs.
mismatches = []
for index, flow_rate in enumerate(flow_readings):
    expected = expected_labels[index]

    actual = buggy_labels[index]
    if actual != expected:
        mismatches.append((flow_rate, expected, actual))

# Test the smallest correction using same cases.
fixed_labels = [fixed_label(value) for value in flow_readings]

# Print concise evidence for learners.
print("Readings:", flow_readings)
print("Expected:", expected_labels)
print("Buggy:", buggy_labels)
print("Mismatch found:", mismatches)
print("Fixed:", fixed_labels)
print("All fixed labels correct:", fixed_labels == expected_labels)

```
