## **1. Variables and Expressions**

### **1.1. Variable Assignment**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1780934752" width="250">



>* Names store values for later use
>* Clear assignments make programs easier to follow

>* Names point to evaluated values
>* Reassignment reflects updated real-world information

>* Meaningful names make programs easier to understand
>* Update assigned values when information changes



```
#@title Python Code - Variable Assignment

# Variable names store reusable engineering values.
# Assignment connects names to current values.
# Later calculations can reuse those names.

# Store shift information with clear names.
shift_hours = 8
units_produced = 960
cycle_time_minutes = 0.45

# Use assigned names in an expression.
units_per_hour = units_produced / shift_hours
labor_minutes = units_produced * cycle_time_minutes

# Update a variable when conditions change.
units_produced = 1000
updated_units_per_hour = units_produced / shift_hours

# Display values and their calculated meaning.
print("Shift hours:", shift_hours)
print("Current units produced:", units_produced)
print("Original cycle time minutes:", cycle_time_minutes)
print("Labor minutes estimated:", labor_minutes)
print("Updated units per hour:", updated_units_per_hour)

```

### **1.2. Naming Variables**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1780934748" width="250">



>* Clear names make Python easier to understand
>* Good variables document engineering calculations

>* Use valid letters, digits, and underscores
>* Choose consistent, descriptive lowercase names

>* Meaningful names reduce confusion and errors
>* Use clear names that remain readable



```
#@title Python Code - Naming Variables

# Clear names make calculations easier.
# Engineering variables should describe meaning.
# Python naming rules prevent confusion.

# Store cycle time using a descriptive name.
cycle_time_minutes = 4.5

# Store demand with clear engineering units.
units_required = 120

# Calculate total production time from variables.
total_time_minutes = cycle_time_minutes * units_required

# Store related names without misleading abbreviations.
target_temperature_celsius = 80
maximum_temperature_celsius = 95

# Compare named values in a readable expression.
temperature_margin_celsius = maximum_temperature_celsius - target_temperature_celsius

# Display results using the variable names.
print("Cycle time per unit:", cycle_time_minutes, "minutes")

# Display the demand stored in a variable.
print("Units required:", units_required, "units")

# Display the calculated production time.
print("Total production time:", total_time_minutes, "minutes")

# Display the safety margin calculation.
print("Temperature margin:", temperature_margin_celsius, "degrees Celsius")

```

### **1.3. Assignment Statements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1780934750" width="250">



>* Store calculated values using meaningful names
>* Evaluate right side before assigning left name

>* Break calculations into clear steps
>* Store results for review and updates

>* Names can receive updated values over time
>* Clear assignments prevent confusion and show logic



```
#@title Python Code - Assignment Statements

# Assignment stores values using readable variable names.
# Expressions calculate results before names receive them.
# Updated assignments reflect changing engineering conditions.

# Store the basic production assumptions.
units_per_hour = 120
operating_hours = 7
planned_downtime_hours = 1

# Calculate available production time first.
productive_hours = operating_hours - planned_downtime_hours
estimated_units = units_per_hour * productive_hours

# Update inventory after production and shipping.
starting_inventory = 450
inventory_after_production = starting_inventory + estimated_units
customer_shipment = 300

# Reassign inventory to the newest known value.
current_inventory = inventory_after_production - customer_shipment

# Display the stored and calculated results.
print("Productive hours:", productive_hours)
print("Estimated units produced:", estimated_units)
print("Current inventory:", current_inventory)

```

## **2. Built In Functions**

### **2.1. Print Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780934754" width="250">



>* Print shows stored results on screen
>* Immediate feedback connects code to outcomes

>* Print intermediate values to check calculations
>* Inspect outputs to find engineering errors

>* Add context to printed results
>* Make outputs clear for decisions



```
#@title Python Code - Print Values

# This example prints engineering values clearly.
# Variables store simple production line information.
# Print statements make calculations visible.

# Store a small production example.
station_name = "Packaging Station A"
units_completed = 120
shift_hours = 8

# Calculate basic performance values.
units_per_hour = units_completed / shift_hours
defect_count = 3
quality_rate = (units_completed - defect_count) / units_completed

# Print raw values with labels.
print("Station:", station_name)
print("Units completed:", units_completed)
print("Shift hours:", shift_hours)

# Print calculated values with context.
print("Units per hour:", units_per_hour)
print("Quality rate:", quality_rate)
print("Defective units:", defect_count)

# Print a readable summary sentence.
summary = f"{station_name} produced {units_per_hour} units per hour."
print(summary)

```

### **2.2. Numeric Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780934756" width="250">



>* Convert text-like numbers for calculations
>* Choose numeric forms Python can use

>* Use integers for counted whole quantities
>* Use floats to preserve measured decimals

>* Clean messy data before converting numbers
>* Choose types carefully for reliable calculations



```
#@title Python Code - Numeric Conversion

# Numeric conversion prepares text for calculations.
# Built in functions inspect data values.
# Industrial examples often mix numeric forms.

# Store raw values like imported records.
units_text = "125"
cycle_time_text = "7.5"
weight_text = "42.8"

# Inspect original Python data types.
print("Original types:", type(units_text).__name__, type(cycle_time_text).__name__)

# Convert counted units into an integer.
units_completed = int(units_text)

# Convert measured values into floating point numbers.
cycle_time_minutes = float(cycle_time_text)
weight_kg = float(weight_text)

# Use converted numbers in engineering calculations.
total_minutes = units_completed * cycle_time_minutes
weight_per_unit = weight_kg / units_completed

# Display converted values and their types.
print("Converted units:", units_completed, type(units_completed).__name__)
print("Cycle time:", cycle_time_minutes, type(cycle_time_minutes).__name__)
print("Total processing minutes:", total_minutes)

# Show rounding only affects displayed results.
print("Weight per unit kg:", round(weight_per_unit, 3))

# Demonstrate careful integer conversion.
truncated_cycle_time = int(cycle_time_minutes)
print("Integer cycle time loses decimal:", truncated_cycle_time)

# Clean a metric text value before conversion.
length_text = "18.2 meters"
length_value = float(length_text.replace(" meters", ""))

# Convert meters to feet for comparison.
length_feet = length_value * 3.28084
print("Length feet:", round(length_feet, 2))

```

### **2.3. Rounding Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780934758" width="250">



>* Round values to useful precision
>* Make engineering results clearer

>* Rounded values are approximations, not exact data
>* Keep precision during calculations; round final results

>* Consistent rounding makes engineering comparisons clearer
>* Choose precision for audience and decision needs



```
#@title Python Code - Rounding Values

# This example demonstrates rounding engineering values.
# Values often need practical displayed precision.
# Keep calculations precise before final reporting.

# Store raw calculated engineering measurements.
temperature_celsius_raw = 73.68472
cycle_time_seconds_raw = 48.23791
material_kg_each_raw = 2.468135
labor_hours_raw = 17.0 / 6.0

# Round values for different reporting needs.
temperature_display = round(temperature_celsius_raw, 1)
cycle_time_display = round(cycle_time_seconds_raw, 2)
material_display = round(material_kg_each_raw, 3)
labor_hours_display = round(labor_hours_raw, 2)

# Round to a whole number when appropriate.
containers_needed_exact = 128.6 / 12.0
containers_display = round(containers_needed_exact)

# Show raw values beside rounded values.
print("Raw temperature:", temperature_celsius_raw)
print("Rounded temperature:", temperature_display)
print("Raw cycle time:", cycle_time_seconds_raw)
print("Rounded cycle time:", cycle_time_display)

# Display practical engineering reporting values.
print("Material per part, kg:", material_display)
print("Labor hours per operator:", labor_hours_display)
print("Estimated containers needed:", containers_display)
print("Original exact containers:", containers_needed_exact)

```

## **3. Engineering Calculations**

### **3.1. Core Engineering Arithmetic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780934759" width="250">



>* Translate engineering situations into arithmetic operations
>* Use Python to calculate meaningful results

>* Operation order changes calculation meaning
>* Use parentheses to group costs clearly

>* Interpret results based on engineering context
>* Round values for practical decisions



```
#@title Python Code - Core Engineering Arithmetic

# Practice arithmetic for industrial engineering calculations.
# Variables store quantities with engineering meaning.
# Parentheses clarify intended calculation order.

# Store production and time assumptions.
setup_minutes = 35
cycle_seconds = 42
planned_units = 120

# Convert cycle time into total minutes.
run_minutes = planned_units * cycle_seconds / 60
job_minutes = setup_minutes + run_minutes

# Convert total job time into hours.
job_hours = job_minutes / 60
throughput_per_hour = planned_units / job_hours

# Store cost assumptions for one production run.
setup_cost = 180
material_cost_each = 4.75
labor_cost_each = 2.10

# Group unit costs before multiplying quantity.
unit_variable_cost = material_cost_each + labor_cost_each
run_cost = setup_cost + planned_units * unit_variable_cost

# Calculate usable output after scrap.
scrap_rate = 0.04
usable_units = planned_units * (1 - scrap_rate)

# Round values for operational reporting.
reported_hours = round(job_hours, 2)
reported_rate = round(throughput_per_hour, 1)
reported_cost = round(run_cost, 2)

# Display concise engineering calculation results.
print("Job time, hours:", reported_hours)
print("Average throughput, units per hour:", reported_rate)
print("Estimated run cost, dollars:", reported_cost)
print("Expected usable units:", round(usable_units, 1))

```

### **3.2. Unit Rate Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780934761" width="250">



>* Compare work using rates per unit
>* Use Python arithmetic to estimate performance

>* Rates turn work into useful performance measures
>* Consistent units support accurate future estimates

>* Keep units consistent before comparing rates
>* Use rates to guide engineering decisions



```
#@title Python Code - Unit Rate Calculations

# Unit rates compare totals against one unit.
# Python arithmetic makes engineering rates transparent.
# Clear units prevent costly calculation mistakes.

# Store shift production and operating time.
parts_completed = 3840
shift_hours = 8
parts_per_hour = parts_completed / shift_hours

# Estimate time needed for a new order.
new_order_parts = 6000
estimated_hours = new_order_parts / parts_per_hour

# Convert minutes before comparing hourly rates.
repair_minutes = 135
machines_repaired = 9
minutes_per_repair = repair_minutes / machines_repaired
repairs_per_hour = 60 / minutes_per_repair

# Calculate material cost using metric units.
material_grams = 2500
kilograms_used = material_grams / 1000
cost_per_kilogram = 4.80
material_cost = kilograms_used * cost_per_kilogram

# Display compact engineering calculation results.
print("Unit Rate Calculations")
print(f"Production rate: {parts_per_hour:.1f} parts per hour")
print(f"Estimated time for order: {estimated_hours:.2f} hours")
print(f"Repair rate: {repairs_per_hour:.2f} machines per hour")
print(f"Material cost: ${material_cost:.2f}")
print("All rates use consistent units before arithmetic.")

```

### **3.3. Unit Rate Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780934763" width="250">



>* Convert totals into useful per-unit rates
>* Compare performance across different operations

>* Choose units carefully for meaningful rates
>* Interpret high rates using process context

>* Use rates to plan resources and schedules
>* Check data quality before making decisions



```
#@title Python Code - Unit Rate Calculations

# Unit rates compare totals against bases.
# Python division calculates engineering rates.
# Clear labels make results understandable.

# Store production totals and shift length.
cartons_completed = 960
shift_hours = 8
cartons_per_hour = cartons_completed / shift_hours

# Store energy use and production volume.
energy_kwh = 420
units_produced = 350
energy_per_unit = energy_kwh / units_produced

# Store distance and fuel consumption.
miles_traveled = 312
fuel_gallons = 12
miles_per_gallon = miles_traveled / fuel_gallons

# Display three useful unit rates.
print("Cartons per hour:", cartons_per_hour)
print("kWh per finished unit:", energy_per_unit)
print("Miles per gallon:", miles_per_gallon)

# Use a rate to estimate future time.
future_order_units = 600
estimated_hours = future_order_units / cartons_per_hour
print("Estimated hours for future order:", estimated_hours)

```
