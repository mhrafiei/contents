## **1. Reading Text Files**

### **1.1. Opening Text Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1780941061" width="250">



>* Open files to connect Python with data
>* Use read mode to protect original records

>* Use accurate file names and paths
>* Know Python’s working folder location

>* Open files only when needed
>* Close them to protect reliable workflows



```
#@title Python Code - Opening Text Files

# Opening files connects programs to stored engineering data.
# Read mode protects original records from accidental changes.
# Context managers close file connections automatically.

# Create a tiny engineering log file.
file_name = "pump_readings.txt"

# Store simple pressure readings as plain text.
sample_lines = ["time_s,pressure_kPa\n", "0,101.3\n", "1,102.1\n", "2,103.0\n"]

# Write the sample file for this demonstration.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.writelines(sample_lines)

# Open the text file safely in read mode.
with open(file_name, "r", encoding="utf-8") as file_object:
    first_line = file_object.readline().strip()
    second_line = file_object.readline().strip()

# Show the opened file contents briefly.
print("Opened file:", file_name)
print("Header line:", first_line)
print("First data record:", second_line)
print("File closed automatically:", file_object.closed)

```

### **1.2. Reading File Contents**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1780941063" width="250">



>* Read file text into program memory
>* Convert text before engineering calculations

>* Read small files as one text block
>* Check memory limits before reading large files

>* Text reading preserves formatting characters
>* Clean, parse, verify, and close files



```
#@title Python Code - Reading File Contents

# Read complete text from a small engineering file.
# Preserve formatting before parsing or calculating values.
# Keep output short for beginner friendly practice.

# Create a tiny pump test text file.
file_name = "pump_test_log.txt"

# Store realistic engineering lines with mixed content.
sample_text = """Pump Test Log
Units: flow=gpm, head=ft
Run,Flow,Head
1,120,58.4
2,135,55.1
3,150,51.8
"""

# Write the text file for this demonstration.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write(sample_text)

# Read the entire file as one string.
with open(file_name, "r", encoding="utf-8") as file_object:
    file_contents = file_object.read()

# Inspect the type and total character count.
print("Object type:", type(file_contents).__name__)
print("Character count:", len(file_contents))

# Show a compact preview of preserved text.
preview_text = file_contents.splitlines()[:3]
print("First three lines:", preview_text)

# Confirm expected engineering labels are present.
has_units_line = "Units:" in file_contents
print("Units line found:", has_units_line)

# Split the full text into separate records.
text_lines = file_contents.splitlines()
print("Number of lines:", len(text_lines))

# Read one numeric field still stored as text.
third_data_line = text_lines[5].split(",")
print("Third flow as text:", third_data_line[1])

```

### **1.3. Line Reading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1780941058" width="250">



>* Process large files one line at a time
>* Extract and inspect each engineering record

>* Lines are read as strings
>* Clean lines and skip non-data records

>* Follow a clear line-by-line workflow
>* Build reliable, testable file handling habits



```
#@title Python Code - Line Reading

# Line reading processes one record at a time.
# Text file values arrive as strings first.
# Engineers clean lines before numeric calculations.

# Create a tiny engineering text file.
file_name = "pump_log.txt"

# Store simple comma separated measurements.
sample_lines = [
    "Pump test log",
    "",
    "time_s,pressure_kpa,flow_lpm",
    "0,101.3,48.2",
    "10,104.8,50.1",
    "20,108.1,51.0",
    "bad line",
    "30,111.4,52.3",
]

# Write the sample lines safely.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write("\n".join(sample_lines))

# Prepare containers for parsed records.
records = []
skipped_lines = 0

# Read one line at a time.
with open(file_name, "r", encoding="utf-8") as file_object:
    for line_number, raw_line in enumerate(file_object, start=1):

        # Remove whitespace and line breaks.
        clean_line = raw_line.strip()

        # Skip headings and empty records.
        if not clean_line or line_number <= 3:
            skipped_lines = skipped_lines + 1
            continue

        # Split comma separated fields.
        parts = clean_line.split(",")

        # Require exactly three useful fields.
        if len(parts) != 3:
            skipped_lines = skipped_lines + 1
            continue

        # Convert fields from text to numbers.
        time_s = float(parts[0])
        pressure_kpa = float(parts[1])
        flow_lpm = float(parts[2])

        # Save the parsed engineering record.
        records.append((time_s, pressure_kpa, flow_lpm))

# Check that useful records were found.
if len(records) == 0:
    raise ValueError("No valid measurement records were parsed.")

# Calculate simple engineering summaries.
average_pressure = sum(row[1] for row in records) / len(records)
average_flow = sum(row[2] for row in records) / len(records)

# Print compact results for learners.
print(f"Valid measurement records: {len(records)}")
print(f"Skipped non-data lines: {skipped_lines}")
print(f"Average pressure: {average_pressure:.1f} kPa")
print(f"Average flow: {average_flow:.1f} L/min")
print(f"First parsed record: {records[0]}")

```

## **2. Parsing File Records**

### **2.1. Splitting Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1780940995" width="250">



>* Split file lines into separate fields
>* Identify each value before calculations

>* Choose the delimiter that matches the file
>* Convert split text fields into numbers

>* Expect messy records and missing fields
>* Validate fields before building data structures



```
#@title Python Code - Splitting Records

# Demonstrate splitting text records safely.
# Engineering files often store delimited measurements.
# Parsed fields become usable Python values.

# Create a tiny comma delimited file.
records_text = "sensor,time_s,temp_C,pressure_kPa\nT01,0.0,22.5,101.3\nT01,1.0,22.8,101.4\nT02,0.0,23.1,100.9\n"

# Save sample records for reading.
file_name = "sensor_records.csv"

# Write the sample text file.
with open(file_name, "w", encoding="utf-8") as file:
    file.write(records_text)

# Read every line from disk.
with open(file_name, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Split the header into field names.
header = lines[0].strip().split(",")

# Prepare storage for parsed dictionaries.
parsed_records = []

# Split each data record carefully.
for line in lines[1:]:
    fields = line.strip().split(",")

    # Skip records with missing fields.
    if len(fields) != len(header):
        continue

    # Convert text measurements into numbers.
    record = {
        "sensor": fields[0],
        "time_s": float(fields[1]),
        "temp_C": float(fields[2]),
        "pressure_kPa": float(fields[3]),
    }

    # Store one cleaned engineering record.
    parsed_records.append(record)

# Compute simple summary values.
avg_temp = sum(r["temp_C"] for r in parsed_records) / len(parsed_records)

# Display compact teaching output.
print("Header fields:", header)
print("First parsed record:", parsed_records[0])
print("Number of valid records:", len(parsed_records))
print("Average temperature, C:", round(avg_temp, 2))

```

### **2.2. Delimited Record Parsing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1780940993" width="250">



>* Delimiters separate text lines into fields
>* Parsed fields become usable structured data

>* Know each record’s expected field structure
>* Convert numeric text for calculations

>* Expect messy real-world delimited data.
>* Validate fields before building data structures.



```
#@title Python Code - Delimited Record Parsing

# Delimited records store measurements in text files.
# Parsing separates fields for engineering calculations.
# Validation catches simple file data problems.

# Create a tiny comma-delimited engineering data file.
file_name = "pump_test_records.csv"

# Keep records small and readable for beginners.
records = [
    "run_id,flow_gpm,head_ft,power_hp",
    "A1,120.0,84.5,6.8",
    "A2,135.5,80.2,7.1",
    "bad_line,missing,value",
    "A3,142.0,78.9,7.4",
]

# Write the sample records to a local file.
with open(file_name, "w", encoding="utf-8") as file:
    file.write("\n".join(records))

# Prepare containers for parsed engineering records.
parsed_records = []
skipped_records = []

# Read each line and split fields by comma.
with open(file_name, "r", encoding="utf-8") as file:
    header = file.readline().strip().split(",")

    # Parse remaining records one line at a time.
    for line_number, line in enumerate(file, start=2):
        fields = line.strip().split(",")

        # Check that each record has four fields.
        if len(fields) != 4:
            skipped_records.append(line_number)
            continue

        # Convert numeric fields from text to floats.
        try:
            run_id = fields[0].strip()
            flow_gpm = float(fields[1])
            head_ft = float(fields[2])
            power_hp = float(fields[3])
        except ValueError:
            skipped_records.append(line_number)
            continue

        # Store one parsed record as a dictionary.
        parsed_records.append({
            "run_id": run_id,
            "flow_gpm": flow_gpm,
            "head_ft": head_ft,
            "power_hp": power_hp,
        })

# Compute a simple summary from parsed records.
average_flow = sum(
    record["flow_gpm"] for record in parsed_records
) / len(parsed_records)

# Display concise results without printing the whole file.
print("Header fields:", header)
print("Parsed record count:", len(parsed_records))
print("Skipped line numbers:", skipped_records)
print("First parsed record:", parsed_records[0])
print("Average flow rate, gpm:", round(average_flow, 1))

```

### **2.3. Header Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1780940997" width="250">



>* Headers describe columns, not measurements
>* Separate headers to interpret data correctly

>* Use headers to name parsed fields
>* Reduce column mix-ups and improve review

>* Clean messy header names before using them
>* Validate headers to catch file problems



```
#@title Python Code - Header Records

# This example separates header records from measurements.
# Header labels become readable dictionary keys.
# Small files keep the parsing process clear.

# Create a tiny comma-delimited engineering data file.
file_name = "pump_test_readings.csv"

# Store one header row and three measurement rows.
sample_text = "time_s, flow_gpm, pressure_psi\n0, 12.5, 48.1\n10, 13.0, 49.4\n20, 12.8, 48.9\n"

# Write the example file using built-in operations.
with open(file_name, "w", encoding="utf-8") as file:
    file.write(sample_text)

# Read the header separately from the data records.
with open(file_name, "r", encoding="utf-8") as file:
    header_line = file.readline()
    data_lines = file.readlines()

# Clean each header field for reliable dictionary keys.
headers = [name.strip() for name in header_line.split(",")]

# Confirm that expected engineering fields are present.
expected = {"time_s", "flow_gpm", "pressure_psi"}
missing = expected.difference(headers)

# Stop early if the header is not usable.
if missing:
    raise ValueError("Missing required header fields")

# Parse each data line into a labeled dictionary.
records = []
for line in data_lines:
    values = [value.strip() for value in line.split(",")]
    record = dict(zip(headers, values))
    records.append(record)

# Convert selected text fields into numerical values.
for record in records:
    record["flow_gpm"] = float(record["flow_gpm"])
    record["pressure_psi"] = float(record["pressure_psi"])

# Use labeled fields for a simple engineering summary.
average_flow = sum(r["flow_gpm"] for r in records) / len(records)
average_pressure = sum(r["pressure_psi"] for r in records) / len(records)

# Display compact results for the learner.
print("Header fields:", headers)
print("First parsed record:", records[0])
print("Average flow, gpm:", round(average_flow, 2))
print("Average pressure, psi:", round(average_pressure, 2))

```

## **3. Writing Summary Reports**

### **3.1. Formatted File Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1780940987" width="250">



>* Save calculated results in clear reports
>* Use labels, units, precision, and organization

>* Clear reports help all readers understand results
>* Consistent formats reduce confusion and errors

>* Choose precision that matches engineering meaning
>* Format files for readers and tools



```
#@title Python Code - Formatted File Output

# Write clear engineering summaries to text files.
# Format values with labels and units.
# Keep reports readable and easy to audit.

from pathlib import Path
from datetime import date

# Store small beam calculation results.
project_name = "Warehouse Beam Check"
beam_span_ft = 24.0

# Define loads and section properties.
uniform_load_lbft = 850.0
max_moment_lbft = uniform_load_lbft * beam_span_ft**2 / 8
section_modulus_in3 = 115.0

# Convert moment units for stress calculation.
max_moment_lbin = max_moment_lbft * 12
bending_stress_psi = max_moment_lbin / section_modulus_in3
allowable_stress_psi = 22000.0

# Calculate a simple utilization ratio.
utilization = bending_stress_psi / allowable_stress_psi
status = "PASS" if utilization <= 1 else "REVIEW"

# Choose a small output file path.
report_path = Path("beam_summary_report.txt")

# Write a formatted engineering summary file.
with report_path.open("w", encoding="utf-8") as file:
    file.write("BEAM DESIGN SUMMARY\n")
    file.write("===================\n")

    file.write(f"Project: {project_name}\n")
    file.write(f"Date: {date.today()}\n\n")

    file.write("Input Data\n")
    file.write(f"  Span:              {beam_span_ft:8.2f} ft\n")

    file.write(f"  Uniform load:      {uniform_load_lbft:8.1f} lb/ft\n")
    file.write(f"  Section modulus:   {section_modulus_in3:8.1f} in^3\n\n")

    file.write("Calculated Results\n")
    file.write(f"  Maximum moment:    {max_moment_lbft:8.1f} lb-ft\n")

    file.write(f"  Bending stress:    {bending_stress_psi:8.0f} psi\n")
    file.write(f"  Allowable stress:  {allowable_stress_psi:8.0f} psi\n")

    file.write(f"  Utilization ratio: {utilization:8.3f}\n")
    file.write(f"  Design status:     {status}\n")

# Read back selected lines for demonstration.
report_lines = report_path.read_text(encoding="utf-8").splitlines()

# Print a compact preview, not the whole file.
print(f"Created file: {report_path}")
print("Preview of formatted report:")

for line_number, report_line in enumerate(report_lines[:10], start=1):
    print(f"{line_number:02d}: {report_line}")

print(f"Total report lines: {len(report_lines)}")

```

### **3.2. Formatted Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1780940989" width="250">



>* Control result layout for clear reports
>* Use labels, spacing, alignment, and precision

>* Match displayed precision to meaningful data
>* Label units and separate value types

>* Consistent reports make repeated results comparable
>* Clear records support review and reuse



```
#@title Python Code - Formatted Output

# Format engineering results for clear reports.
# Use labels, alignment, and precision.
# Write a small summary text file.

# Store small beam calculation results.
beam_results = [
    {"beam": "B1", "span_m": 4.5, "deflection_mm": 7.236, "stress_mpa": 142.82, "safety_factor": 2.138},

    {"beam": "B2", "span_m": 6.0, "deflection_mm": 12.894, "stress_mpa": 188.47, "safety_factor": 1.617},

    {"beam": "B3", "span_m": 5.2, "deflection_mm": 9.105, "stress_mpa": 165.03, "safety_factor": 1.845},
]

# Choose a concise output filename.
report_file = "beam_summary_report.txt"

# Validate that records are available.
if len(beam_results) == 0:
    raise ValueError("No beam results were provided.")

# Write aligned headings and formatted values.
with open(report_file, "w", encoding="utf-8") as file:
    file.write("Beam Deflection Summary Report\n")
    file.write("Generated from built-in Python formatting\n")

    file.write("\n")
    file.write(f"{'Beam':<8}{'Span (m)':>12}{'Defl. (mm)':>14}{'Stress (MPa)':>15}{'SF':>8}\n")
    file.write("-" * 57 + "\n")

    for item in beam_results:
        line = (
            f"{item['beam']:<8}"
            f"{item['span_m']:>12.2f}"
            f"{item['deflection_mm']:>14.1f}"
            f"{item['stress_mpa']:>15.1f}"
            f"{item['safety_factor']:>8.2f}\n"
        )

        file.write(line)

# Read back a short preview.
with open(report_file, "r", encoding="utf-8") as file:
    preview_lines = file.readlines()[:7]

# Display only a compact report preview.
print("Formatted report preview:")
print("".join(preview_lines))
print(f"Report saved as: {report_file}")

```

### **3.3. Structured Summary Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1780940991" width="250">



>* Record inputs, calculations, and results clearly
>* Organize reports for review and communication

>* Organize reports into clear, consistent sections
>* Use labels, spacing, units, and logical flow

>* Consistent reports make repeated calculations comparable
>* Clear labels prevent confusion and support decisions



```
#@title Python Code - Structured Summary Files

# Create a structured engineering summary file.
# Use labels, sections, units, and dates.
# Keep reports readable for later review.

from datetime import date
from pathlib import Path

# Store one simple beam calculation case.
case_name = "Beam Case A"
beam_length_ft = 12.0
load_lb = 850.0
elastic_modulus_ksi = 29000.0

# Record calculated engineering outputs with units.
max_deflection_in = 0.184
allowable_deflection_in = beam_length_ft * 12.0 / 360.0
factor_of_safety = allowable_deflection_in / max_deflection_in

# Choose a small report file location.
report_path = Path("beam_summary_report.txt")

# Build clear, structured report sections.
report_lines = [
    "BEAM DEFLECTION SUMMARY REPORT",
    f"Analysis date: {date.today()}",
    f"Case name: {case_name}",
    "",

    "INPUT PARAMETERS",
    f"Beam length: {beam_length_ft:.1f} ft",
    f"Point load: {load_lb:.1f} lb",
    f"Elastic modulus: {elastic_modulus_ksi:.0f} ksi",

    "",
    "CALCULATED RESULTS",
    f"Maximum deflection: {max_deflection_in:.3f} in",
    f"Allowable deflection: {allowable_deflection_in:.3f} in",
    f"Deflection safety factor: {factor_of_safety:.2f}",

    "",
    "CONCLUSION",
    "Design passes deflection limit.",
]

# Write the structured report using built-in operations.
with report_path.open("w", encoding="utf-8") as file_object:
    file_object.write("\n".join(report_lines))

# Read back a few lines to verify formatting.
with report_path.open("r", encoding="utf-8") as file_object:
    preview_lines = file_object.readlines()[:8]

# Print a compact preview for learners.
print("Structured report file created:", report_path)
print("Preview of first report lines:")
print("".join(preview_lines))

```
