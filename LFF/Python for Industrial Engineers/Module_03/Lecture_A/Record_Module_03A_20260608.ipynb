## **1. Function Design**

### **1.1. Define Function Purpose**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780939963" width="250">



>* Give each function one clear calculation purpose
>* Build small reusable functions for larger workflows

>* Identify required inputs and expected outputs
>* Keep each function focused on one concept

>* Clear purpose prevents team misunderstanding.
>* Defined scope makes functions trustworthy and reusable.



```
#@title Python Code - Define Function Purpose

# Function purpose guides focused engineering calculations.
# Parameters define required input information clearly.
# Return values make results reusable later.

def calculate_labor_productivity(units_completed, labor_hours):
    # Validate inputs before performing calculation.
    if units_completed < 0 or labor_hours <= 0:
        raise ValueError("Use nonnegative units and positive hours")

    # Return one focused productivity measure.
    productivity = units_completed / labor_hours
    return productivity

# Store representative shift input values.
day_units, day_hours = 960, 80
night_units, night_hours = 810, 75

# Reuse the same focused function twice.
day_productivity = calculate_labor_productivity(day_units, day_hours)
night_productivity = calculate_labor_productivity(night_units, night_hours)

# Print concise engineering interpretation.
print("Function purpose: calculate labor productivity only")
print(f"Day shift: {day_productivity:.2f} units per labor hour")
print(f"Night shift: {night_productivity:.2f} units per labor hour")
print("Same purpose, different parameters, reusable result")

```

### **1.2. Scope Metrics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780939961" width="250">



>* Define one clear engineering metric per function
>* Make repeated measurements reusable across operations

>* Keep calculations focused on one metric
>* Separate inputs from unrelated program tasks

>* Choose metrics with clear operational meaning
>* Reduce ambiguity for consistent engineering decisions



```
#@title Python Code - Scope Metrics

# Functions should answer one operational question.
# Scope metrics keep calculations focused and reusable.
# This example separates two warehouse measures.

# Define one function for picking productivity.
def picking_productivity(order_lines, labor_hours):
    """Return order lines completed per labor hour."""
    if labor_hours <= 0:
        raise ValueError("labor_hours must be positive")

    return order_lines / labor_hours

# Define one function for defect percentage.
def defect_percentage(defective_units, total_units):
    """Return defective units as a percentage."""
    if total_units <= 0:
        raise ValueError("total_units must be positive")

    return defective_units / total_units * 100

# Store representative engineering inputs.
order_lines_completed = 360
labor_hours_worked = 8
bad_units_found = 14
units_inspected = 700

# Calculate each scoped metric separately.
lines_per_hour = picking_productivity(
    order_lines_completed, labor_hours_worked
)

# Calculate defect rate using separate inputs.
defect_rate = defect_percentage(
    bad_units_found, units_inspected
)

# Print concise results for interpretation.
print("Scope metric: picking productivity")
print(f"Result: {lines_per_hour:.1f} order lines per labor hour")
print("Scope metric: defect percentage")
print(f"Result: {defect_rate:.2f}% defective units")

```

### **1.3. Measurement Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780939965" width="250">



>* Convert observations into clear engineering metrics
>* Keep calculations focused, reusable, and verifiable

>* Define each metric and required inputs clearly
>* Avoid assumptions to prevent shared calculation errors

>* Use one trusted calculation for each metric
>* Keep functions focused, reusable, and transparent



```
#@title Python Code - Measurement Functions

# Measurement functions convert observations into engineering metrics.
# Each function answers one focused calculation question.
# Return values make calculations reusable across cases.

# Define a cycle time measurement function.
def cycle_time_minutes(start_minute, finish_minute):
    duration = finish_minute - start_minute
    return duration

# Define a utilization measurement function.
def utilization_percent(busy_minutes, available_minutes):
    if available_minutes <= 0:
        return None

    utilization = busy_minutes / available_minutes * 100
    return utilization

# Define a defect rate measurement function.
def defect_rate_percent(defective_units, inspected_units):
    if inspected_units <= 0:
        return None

    defect_rate = defective_units / inspected_units * 100
    return defect_rate

# Define a throughput measurement function.
def throughput_per_hour(completed_units, elapsed_minutes):
    if elapsed_minutes <= 0:
        return None

    hours = elapsed_minutes / 60
    return completed_units / hours

# Store representative engineering observations.
start_time = 15
finish_time = 22
busy_time = 410
available_time = 480

defects_found = 7
units_checked = 350
units_completed = 96
elapsed_time = 120

# Calculate each metric using reusable functions.
cycle_time = cycle_time_minutes(start_time, finish_time)
utilization = utilization_percent(busy_time, available_time)
defect_rate = defect_rate_percent(defects_found, units_checked)
throughput = throughput_per_hour(units_completed, elapsed_time)

# Display concise engineering measurement results.
print("Reusable measurement function results:")
print(f"Cycle time: {cycle_time:.1f} minutes per unit")
print(f"Utilization: {utilization:.1f}% of available time")
print(f"Defect rate: {defect_rate:.2f}% of inspected units")
print(f"Throughput: {throughput:.1f} units per hour")

# Test one function with another workstation case.
second_cycle = cycle_time_minutes(30, 39)
print(f"Second workstation cycle time: {second_cycle:.1f} minutes")

```

## **2. Parameters and Returns**

### **2.1. Function Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780939973" width="250">



>* Parameters pass case-specific values into functions
>* One calculation adapts across operating conditions

>* Choose parameters that may vary
>* Clear inputs reveal calculation assumptions

>* Avoid duplicate calculations and manual edits
>* Compare scenarios with consistent function logic



```
#@title Python Code - Function Parameters

# Functions reuse calculations across cases.
# Parameters carry changing engineering inputs.
# Return values support later analysis.

# Define a capacity function using three parameters.
def estimate_units(shift_hours, cycle_minutes, efficiency):
    # Convert available hours into productive minutes.
    productive_minutes = shift_hours * 60 * efficiency

    # Return completed units for this case.
    return productive_minutes / cycle_minutes

# Store representative machine cases.
weekday_units = estimate_units(8, 2.5, 0.90)
weekend_units = estimate_units(6, 2.5, 0.85)
peak_units = estimate_units(10, 2.2, 0.92)

# Print each returned result clearly.
print(f"Weekday capacity: {weekday_units:.1f} units")
print(f"Weekend capacity: {weekend_units:.1f} units")
print(f"Peak capacity: {peak_units:.1f} units")

# Compare scenarios using returned values.
extra_units = peak_units - weekday_units
print(f"Peak adds {extra_units:.1f} units over weekday")

```

### **2.2. Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1780939975" width="250">



>* Return values send results back
>* Results support flexible engineering calculations

>* Keep calculations separate from display choices
>* Returned results support many later uses

>* Return one or multiple useful results
>* Design outputs for testing and reuse



```
#@title Python Code - Return Values

# Functions can send useful results back.
# Returned values support reusable engineering calculations.
# This example separates calculation from printing.

# Define a focused processing-time function.
def batch_processing_hours(part_count, cycle_seconds, setup_minutes):
    # Validate realistic positive engineering inputs.
    if part_count < 0 or cycle_seconds < 0 or setup_minutes < 0:
        raise ValueError("Inputs must be nonnegative values.")

    # Convert cycle work into hours.
    run_hours = (part_count * cycle_seconds) / 3600
    # Convert setup work into hours.
    setup_hours = setup_minutes / 60

    # Return a reusable numeric result.
    return run_hours + setup_hours

# Call the function for two batches.
batch_a_hours = batch_processing_hours(240, 45, 30)
batch_b_hours = batch_processing_hours(180, 50, 20)

# Reuse returned values in later analysis.
combined_hours = batch_a_hours + batch_b_hours
shift_capacity_hours = 8

# Compare returned results with capacity.
remaining_hours = shift_capacity_hours - combined_hours
feasible_schedule = combined_hours <= shift_capacity_hours

# Print a compact supervisor summary.
print("Batch A hours:", round(batch_a_hours, 2))
print("Batch B hours:", round(batch_b_hours, 2))
print("Combined hours:", round(combined_hours, 2))
print("Remaining shift hours:", round(remaining_hours, 2))
print("Schedule feasible:", feasible_schedule)

```

### **2.3. Default Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780939976" width="250">



>* Defaults handle missing or unusable inputs
>* Predictable returns keep functions reusable

>* Defaults flag invalid or boundary inputs
>* Prevent misleading results in later analysis

>* Choose defaults that preserve calculation meaning
>* Test edge cases for reliable reuse



```
#@title Python Code - Default Return Values

# Functions can return safe default values.
# Defaults protect reusable engineering calculations.
# Edge cases need explicit testing.

def average_cycle_time(times, default=None):
    # Reject missing or empty cycle lists.
    if not times:
        return default

    # Keep only positive numeric observations.
    valid_times = [time for time in times if time > 0]

    # Return default when no times remain.
    if not valid_times:
        return default

    # Return the reusable average calculation.
    return sum(valid_times) / len(valid_times)

# Define representative workstation examples.
station_a_times = [42, 39, 45]
station_b_times = []
station_c_times = [-5, 0]

# Calculate results using the same function.
a_average = average_cycle_time(station_a_times, default="Unavailable")
b_average = average_cycle_time(station_b_times, default="Unavailable")
c_average = average_cycle_time(station_c_times, default="Unavailable")

# Display normal and default return behavior.
print("Station A average seconds:", a_average)
print("Station B average seconds:", b_average)
print("Station C average seconds:", c_average)
print("Default values prevent misleading zero cycle times.")

```

## **3. Testing Documented Functions**

### **3.1. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780939967" width="250">



>* Test functions with realistic engineering examples
>* Confirm outputs match documented expectations

>* Test typical, boundary, and unusual valid cases
>* Check assumptions about units, percentages, and zeros

>* Match tests to documented real-world uses
>* Verify expected results to build reliability



```
#@title Python Code - Representative Test Cases

# Representative tests connect documentation with evidence.
# Each case uses realistic engineering inputs.
# Expected values are known before execution.

# Define a documented utilization function.
def workstation_utilization(cycle_minutes, hourly_demand, available_minutes):
    """Return utilization as a decimal for one workstation."""

    # Validate physically meaningful engineering inputs.
    if cycle_minutes < 0 or hourly_demand < 0:
        raise ValueError("cycle_minutes and hourly_demand must be nonnegative")

    # Prevent division by impossible available time.
    if available_minutes <= 0:
        raise ValueError("available_minutes must be positive")

    # Convert hourly demand into shift workload minutes.
    workload_minutes = cycle_minutes * hourly_demand * 8

    # Return utilization for the available shift time.
    return workload_minutes / available_minutes

# Create representative documented test cases.
test_cases = [
    ("typical shift", 3.0, 40, 480, 0.75),
    ("full capacity", 4.0, 15, 480, 1.00),
    ("low volume", 12.0, 2, 480, 0.40),
]

# Check each case against its expected result.
results = []

# Run compact tests with manual expected values.
for name, cycle, demand, minutes, expected in test_cases:
    actual = workstation_utilization(cycle, demand, minutes)

    # Compare rounded values for beginner readability.
    passed = round(actual, 2) == round(expected, 2)
    results.append((name, actual, expected, passed))

# Print a concise test report.
print("Representative utilization tests:")

# Display each case without excessive output.
for name, actual, expected, passed in results:
    status = "PASS" if passed else "CHECK"

    # Format utilization values as percentages.
    actual_pct = f"{actual:.0%}"
    expected_pct = f"{expected:.0%}"

    # Show case name, output, expectation, and result.
    print(f"{name}: actual {actual_pct}, expected {expected_pct}, {status}")

# Summarize whether every representative case passed.
all_passed = all(item[3] for item in results)
print(f"All representative tests passed: {all_passed}")

```

### **3.2. Representative Input Tests**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780939971" width="250">



>* Test functions with realistic engineering inputs
>* Check calculations, units, and expected ranges

>* Test typical, boundary, and misunderstanding cases
>* Check assumptions, units, and documentation clarity

>* Test functions across realistic future uses
>* Pair balanced tests with clear documentation



```
#@title Python Code - Representative Input Tests

# Representative tests check documented engineering functions.
# Realistic inputs reveal unit misunderstandings early.
# Small cases keep results easy to inspect.

# Define a documented utilization calculation function.
def calculate_utilization(operating_hours, available_hours):
    """Return utilization as operating hours divided by available hours."""

    # Guard against impossible available time.
    if available_hours <= 0:
        raise ValueError("available_hours must be positive")

    # Return a reusable decimal result.
    return operating_hours / available_hours

# Store representative engineering input cases.
test_cases = [
    ("Lightly loaded press", 2.5, 8.0, "typical"),
    ("Busy packaging line", 6.8, 8.0, "typical"),

    ("Near full capacity", 7.9, 8.0, "boundary"),
    ("Zero operating time", 0.0, 8.0, "boundary"),
]

# Print a compact heading for inspection.
print("Representative input tests for documented utilization function")

# Test each realistic case clearly.
for name, operating, available, category in test_cases:
    result = calculate_utilization(operating, available)

    # Format each result as a percent.
    percent_text = f"{result:.1%} utilization"
    print(f"{category}: {name} -> {percent_text}")

# Test a documented misunderstanding case safely.
try:
    calculate_utilization(420, 8)

# Explain why this plausible input looks suspicious.
except ValueError as error:
    print("Invalid case caught:", error)
else:
    print("Unit warning: 420 minutes used where hours were expected")

```

### **3.3. Representative Input Tests**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780939969" width="250">



>* Test functions with realistic engineering inputs
>* Check outputs against professional expectations

>* Test ordinary, boundary, and unusual inputs
>* Build trust across real system variation

>* Explain why test inputs were chosen
>* Connect outputs to engineering judgment



```
#@title Python Code - Representative Input Tests

# Representative tests connect documentation with engineering reality.
# Focused functions should be checked using realistic inputs.
# Short assertions confirm expected behavior without noisy output.

# Define a documented industrial engineering function.
def labor_hours(task_minutes, batch_units, workers):
    """Return labor hours for a batch production task."""

    # Validate inputs before calculation.
    if task_minutes < 0 or batch_units < 0:
        raise ValueError("task_minutes and batch_units must be nonnegative")

    # Validate worker count separately.
    if workers <= 0:
        raise ValueError("workers must be positive")

    # Convert total worker minutes into hours.
    total_minutes = task_minutes * batch_units
    return total_minutes / (60 * workers)

# Store representative engineering test cases.
tests = [
    ("ordinary", 2.5, 120, 3, 1.6667),
    ("boundary", 4.0, 0, 2, 0.0),
    ("unusual", 1.2, 1000, 5, 4.0),
]

# Check each documented representative input.
passed_tests = 0
for name, minutes, units, workers, expected in tests:

    # Calculate and compare within rounding tolerance.
    result = labor_hours(minutes, units, workers)
    passed = abs(result - expected) < 0.001

    # Count successful tests for a concise summary.
    if passed:
        passed_tests = passed_tests + 1

    # Print one engineering interpretation per case.
    print(f"{name}: {result:.2f} labor hours, expected {expected:.2f}")

# Verify all representative cases matched expectations.
print(f"Representative tests passed: {passed_tests} of {len(tests)}")

```
