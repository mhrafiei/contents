## **1. Workflow Design**

### **1.1. Input Identification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_01_01.jpg?v=1780944457" width="250">



>* Reliable automation starts with complete input discovery
>* Map every human lookup before coding

>* Trace triggers to required information sources
>* Treat each separate source as input

>* Define input quality, detail, and limits
>* Flag required fields and review exceptions



```
#@title Python Code - Input Identification

# Identify inputs before workflow automation.
# Use small records for clarity.
# Validate availability before processing.

# Store expected workflow input definitions.
required_inputs = {
    "orders": "daily customer orders file",
    "rates": "standard units per hour",
    "downtime": "machine downtime minutes",
}

# Store received inputs for one report.
received_inputs = {
    "orders": [{"line": "A", "units": 420}],
    "rates": {"A": 60, "B": 55},
    "downtime": {"A": 35, "B": None},
}

# Check whether each required source exists.
missing_sources = []
for key_name in required_inputs:
    if key_name not in received_inputs:
        missing_sources.append(key_name)

# Check values that could break calculations.
problem_notes = []
for line_name, rate_value in received_inputs["rates"].items():
    if not isinstance(rate_value, (int, float)):
        problem_notes.append(line_name + " rate is not numeric")

# Check downtime values before capacity calculations.
for line_name, minutes_value in received_inputs["downtime"].items():
    if minutes_value is None:
        problem_notes.append(line_name + " downtime is missing")

# Summarize identified input readiness.
print("Workflow input identification check")
print("Required sources:", len(required_inputs))
print("Received sources:", len(received_inputs))
print("Missing source count:", len(missing_sources))
print("Validation issue count:", len(problem_notes))

# Print limited review notes for learners.
for note_text in problem_notes[:3]:
    print("Review needed:", note_text)

# Decide whether automation should continue.
ready_to_continue = not missing_sources and not problem_notes
print("Ready for automated calculations:", ready_to_continue)

```

### **1.2. Process Steps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_01_02.jpg?v=1780944459" width="250">



>* Turn manual work into ordered script steps
>* Break broad tasks into testable actions

>* Make every workflow decision explicit
>* Define rules before writing Python scripts

>* Stage workflows from data to decisions
>* Document choices before automating them



```
#@title Python Code - Process Steps

# Design workflow steps before automating operations.
# Each step transforms or checks information.
# Small examples reveal hidden decision rules.

# Store tiny production records as dictionaries.
records = [
    {"date": "2026-06-01", "line": "A", "planned": 480, "actual": 438},
    {"date": "2026-06-01", "line": "B", "planned": 450, "actual": 463},
    {"date": "2026-06-02", "line": "A", "planned": 480, "actual": None},
]

# Define the explicit workflow stages.
workflow_steps = [
    "prepare records for selected date",
    "validate required production fields",
    "calculate efficiency from planned and actual",
    "flag lines below the target threshold",
]

# Choose simple workflow parameters.
target_date = "2026-06-01"
target_efficiency = 0.95
required_fields = {"date", "line", "planned", "actual"}

# Filter records during preparation.
selected_records = [
    row for row in records if row.get("date") == target_date
]

# Validate that matching records exist.
if len(selected_records) == 0:
    raise ValueError("No records matched the target date.")

# Validate fields before calculation.
for row in selected_records:
    missing_fields = required_fields - set(row)
    if missing_fields:
        raise ValueError("A selected record is missing fields.")

# Transform validated records into results.
results = []
for row in selected_records:
    planned_output = row["planned"]
    actual_output = row["actual"]

    # Check values before division.
    if not isinstance(planned_output, (int, float)):
        raise TypeError("Planned output must be numeric.")
    if planned_output <= 0:
        raise ValueError("Planned output must be positive.")

    # Calculate efficiency for one line.
    efficiency = actual_output / planned_output
    status = "review" if efficiency < target_efficiency else "acceptable"

    # Save organized result for reporting.
    results.append(
        {"line": row["line"], "efficiency": efficiency, "status": status}
    )

# Print a compact process design summary.
print("Process steps:")
for step_number, step_name in enumerate(workflow_steps, start=1):
    print(f"{step_number}. {step_name}")

# Print concise workflow results.
print("Results:")
for item in results:
    percent = item["efficiency"] * 100
    print(f"Line {item['line']}: {percent:.1f}% efficiency, {item['status']}")

```

### **1.3. Output Definition**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_01_03.jpg?v=1780944461" width="250">



>* Define the exact workflow result first
>* Use clear outputs to guide automation

>* Match output format to each audience
>* Define outputs early to avoid confusion

>* Show result confidence and review needs
>* Make outputs clear, useful, and actionable



```
#@title Python Code - Output Definition

# Define outputs before building workflow steps.
# Keep automation focused on user decisions.
# Validate results before sharing recommendations.

import csv
from pathlib import Path

# Store tiny production records as dictionaries.
production_rows = [
    {"line": "A", "shift": "Day", "planned": 500, "actual": 475, "scrap": 12, "downtime": 35},
    {"line": "B", "shift": "Day", "planned": 420, "actual": 438, "scrap": 8, "downtime": 18},
    {"line": "C", "shift": "Night", "planned": 390, "actual": 355, "scrap": 22, "downtime": 64},
]

# Define the exact output columns first.
output_columns = [
    "line", "shift", "planned", "actual",
    "scrap", "downtime", "met_target", "review_flag",
]

# Validate records before calculating output fields.
def validate_record(record):
    required_keys = {"line", "shift", "planned", "actual", "scrap", "downtime"}
    missing_keys = required_keys.difference(record.keys())
    return not missing_keys and record["planned"] > 0

# Convert each input record into decision-ready output.
def build_output_row(record):
    met_target = record["actual"] >= record["planned"]
    review_flag = record["scrap"] > 20 or record["downtime"] > 45
    return {
        "line": record["line"], "shift": record["shift"],
        "planned": record["planned"], "actual": record["actual"],
        "scrap": record["scrap"], "downtime": record["downtime"],
        "met_target": met_target, "review_flag": review_flag,
    }

# Build output only from valid records.
valid_rows = [row for row in production_rows if validate_record(row)]
output_rows = [build_output_row(row) for row in valid_rows]

# Save a small decision-ready summary file.
output_path = Path("shift_output_definition.csv")
with output_path.open("w", newline="") as file_handle:
    writer = csv.DictWriter(file_handle, fieldnames=output_columns)
    writer.writeheader()
    writer.writerows(output_rows)

# Print concise results for the intended user.
print("Defined output: shift production summary CSV")
print(f"Rows created: {len(output_rows)}")
print(f"Columns: {', '.join(output_columns)}")
print(f"Saved file: {output_path.name}")
print("Lines needing review:", sum(row["review_flag"] for row in output_rows))

```

## **2. Script Integration**

### **2.1. Function File Integration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_02_01.jpg?v=1780944465" width="250">



>* Connect functions with input and output files
>* Automate repeatable work with fewer errors

>* Separate file handling from data analysis
>* Test functions before full workflow integration

>* Prepare consistent data before analysis
>* Sequence functions to protect reliable outputs



```
#@title Python Code - Function File Integration

# Integrate functions with workflow files.
# Validate records before calculating summaries.
# Save dependable outputs for review.

import csv
from pathlib import Path

# Create small workflow input paths.
input_path = Path("daily_production.csv")
output_path = Path("line_summary.csv")

# Store tiny production records safely.
records = [
    {"line": "A", "planned": "100", "actual": "96"},
    {"line": "B", "planned": "120", "actual": "126"},
    {"line": "C", "planned": "0", "actual": "75"},
]

# Write sample records like exported data.
with input_path.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=records[0].keys())
    writer.writeheader()
    writer.writerows(records)

# Convert text fields into integers.
def clean_record(row):
    planned = int(row["planned"])
    actual = int(row["actual"])
    return {"line": row["line"], "planned": planned, "actual": actual}

# Validate common calculation risks early.
def validate_record(record):
    if record["planned"] <= 0:
        return False
    if record["actual"] < 0:
        return False
    return True

# Calculate efficiency only after validation.
def efficiency(record):
    value = record["actual"] / record["planned"]
    return round(value * 100, 1)

# Read, clean, validate, and summarize file data.
valid_rows = []
skipped_lines = []
with input_path.open("r", newline="") as file:
    reader = csv.DictReader(file)

    for row in reader:
        cleaned = clean_record(row)
        if validate_record(cleaned):
            cleaned["efficiency_percent"] = efficiency(cleaned)
            valid_rows.append(cleaned)
        else:
            skipped_lines.append(cleaned["line"])

# Confirm useful output before writing results.
if not valid_rows:
    raise ValueError("No valid production rows were available.")

# Save integrated workflow results.
fieldnames = list(valid_rows[0].keys())
with output_path.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(valid_rows)

# Report concise workflow results.
average_efficiency = sum(row["efficiency_percent"] for row in valid_rows)
average_efficiency = round(average_efficiency / len(valid_rows), 1)
print(f"Valid lines summarized: {len(valid_rows)}")
print(f"Skipped invalid lines: {skipped_lines}")
print(f"Average efficiency: {average_efficiency}%")
print(f"Output file created: {output_path.name}")

```

### **2.2. Function File Integration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_02_02.jpg?v=1780944463" width="250">



>* Files supply changing workplace data
>* Functions organize repeatable processing steps

>* Store file data in useful collections
>* Separate reading logic from analysis functions

>* Clear function roles improve workflow reliability
>* Structured data flow supports adaptable automation



```
#@title Python Code - Function File Integration

# Integrate functions with small workflow files.
# Validate records before calculating production metrics.
# Keep each workflow step separately testable.

import csv
from pathlib import Path

# Create a tiny machine status input file.
input_path = Path("machine_status.csv")
output_path = Path("downtime_summary.csv")

# Store sample records as dictionaries.
sample_rows = [
    {"machine": "Press_1", "minutes": "18", "reason": "setup"},
    {"machine": "Press_1", "minutes": "12", "reason": "maintenance"},

    {"machine": "Cutter_2", "minutes": "0", "reason": "none"},
    {"machine": "Cutter_2", "minutes": "27", "reason": "material"},
]

# Write the sample input file.
with input_path.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=sample_rows[0].keys())
    writer.writeheader()
    writer.writerows(sample_rows)

# Read records from a CSV file.
def load_records(path):
    if not path.exists():
        return []

    with path.open("r", newline="") as file:
        return list(csv.DictReader(file))

# Validate and standardize one record.
def clean_record(row):
    required = {"machine", "minutes", "reason"}
    if not required.issubset(row):
        return None

    try:
        minutes = float(row["minutes"])
    except ValueError:
        return None

    if minutes < 0 or row["machine"] == "":
        return None

    return {"machine": row["machine"], "minutes": minutes, "reason": row["reason"]}

# Summarize downtime minutes by machine.
def summarize_downtime(records):
    summary = {}
    for record in records:
        machine = record["machine"]
        summary[machine] = summary.get(machine, 0) + record["minutes"]

    return summary

# Save the summary for later review.
def save_summary(summary, path):
    with path.open("w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["machine", "downtime_minutes"])

        for machine, minutes in sorted(summary.items()):
            writer.writerow([machine, minutes])

# Run the integrated workflow steps.
raw_records = load_records(input_path)
clean_records = [clean_record(row) for row in raw_records]
valid_records = [row for row in clean_records if row is not None]
summary = summarize_downtime(valid_records)

# Persist and report compact results.
save_summary(summary, output_path)
print(f"Input records read: {len(raw_records)}")
print(f"Valid records used: {len(valid_records)}")
print(f"Machines summarized: {len(summary)}")
print(f"Summary file created: {output_path.name}")

# Show the key workflow result.
print(f"Downtime summary: {summary}")

```

### **2.3. Conditional Workflow Logic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_02_03.jpg?v=1780944467" width="250">



>* Conditionals guide automated workflow decisions
>* Handle incomplete or changing operational data

>* Conditionals connect files, functions, and collections
>* Workflows adapt to real operational decisions

>* Make decision points clear and purposeful
>* Use conditionals to guide reliable workflows



```
#@title Python Code - Conditional Workflow Logic

# Conditional logic connects workflow pieces.
# Small files can simulate operations data.
# Validation prevents unsafe automated decisions.

import csv
from pathlib import Path

# Create a small production report file.
data_path = Path("daily_output.csv")
rows = [
    ["Station", "Planned", "Actual", "ShiftHours"],

    ["Cutting", "120", "114", "8"],
    ["Assembly", "150", "", "8"],
    ["Packing", "100", "92", "0"],

    ["Inspection", "80", "78", "8"],
]

# Write the tiny source file.
with data_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Convert safe numeric text values.
def safe_float(text_value, field_name):
    if text_value == "":
        return None, f"missing {field_name}"

    try:
        return float(text_value), ""
    except ValueError:
        return None, f"invalid {field_name}"

# Classify station output performance.
def classify_station(planned, actual, hours):
    if planned <= 0 or hours <= 0:
        return "exception", "nonpositive planned or hours"

    utilization = actual / planned
    if utilization >= 0.95:
        return "normal", "meets target"

    if utilization >= 0.85:
        return "review", "near target"
    return "bottleneck", "below target"

# Process records using conditional workflow logic.
results = []
warnings = []
with data_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    for record in reader:

        planned, planned_error = safe_float(record["Planned"], "planned")
        actual, actual_error = safe_float(record["Actual"], "actual")
        hours, hours_error = safe_float(record["ShiftHours"], "hours")

        errors = [planned_error, actual_error, hours_error]
        errors = [error for error in errors if error]

        if errors:
            warnings.append((record["Station"], "; ".join(errors)))
            continue

        status, reason = classify_station(planned, actual, hours)
        results.append((record["Station"], status, reason))

# Write valid workflow results only.
output_path = Path("workflow_summary.csv")
with output_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Station", "Status", "Reason"])

    writer.writerows(results)

# Show a concise automation summary.
print(f"Source file processed: {data_path.name}")
print(f"Valid station records: {len(results)}")
print(f"Warning records skipped: {len(warnings)}")
print("Workflow decisions:")

for station, status, reason in results:
    print(f"- {station}: {status} ({reason})")

if warnings:
    first_station, first_warning = warnings[0]
    print(f"First warning: {first_station} has {first_warning}")

print(f"Summary file written: {output_path.name}")

```

## **3. Validation Checks**

### **3.1. Input Validation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_03_01.jpg?v=1780944468" width="250">



>* Check inputs before the workflow runs
>* Prevent misleading results and build trust

>* Check input structure matches workflow expectations
>* Confirm values are realistic and actionable

>* Clear errors guide users to fixes
>* Early validation builds trustworthy workflows



```
#@title Python Code - Input Validation

# Validate production inputs before workflow calculations.
# Small records represent submitted shift reports.
# Clear messages help users fix data.

# Import pandas for tabular validation.
import pandas as pd

# Create tiny production data with intentional issues.
records = [
    {"shift": "A", "machine": "M1", "units": 120, "scrap": 3},
    {"shift": "B", "machine": "M2", "units": 80, "scrap": 95},
    {"shift": "C", "machine": "", "units": -10, "scrap": 1},
]

# Convert records into a small DataFrame.
df = pd.DataFrame(records)

# Define the required workflow input fields.
required = ["shift", "machine", "units", "scrap", "downtime_minutes"]

# Collect all validation messages here.
issues = []

# Check whether every expected column exists.
missing = [col for col in required if col not in df.columns]

# Record missing structural input problems.
if missing:
    issues.append("Missing columns: " + ", ".join(missing))

# Add a safe default for demonstration.
if "downtime_minutes" not in df.columns:
    df["downtime_minutes"] = 0

# Check for blank machine names.
blank_machine = df["machine"].astype(str).str.strip().eq("")

# Report blank machine rows clearly.
if blank_machine.any():
    issues.append("Blank machine name in row " + str(blank_machine.idxmax()))

# Check for negative production quantities.
negative_units = df["units"] < 0

# Report impossible production quantities.
if negative_units.any():
    issues.append("Negative units in row " + str(negative_units.idxmax()))

# Check scrap does not exceed production.
excess_scrap = df["scrap"] > df["units"]

# Report meaning based validation failures.
if excess_scrap.any():
    issues.append("Scrap exceeds units in row " + str(excess_scrap.idxmax()))

# Show a compact validation result.
print("Validation summary:")

# Limit printed issues for readable output.
for message in issues[:5]:
    print("- " + message)

# Continue only when validation passes.
if not issues:
    print("Inputs passed; workflow calculations may begin.")
else:
    print("Fix inputs before calculating performance metrics.")

```

### **3.2. Input Validation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_03_02.jpg?v=1780944470" width="250">



>* Check inputs before automated workflows use them
>* Block bad data to prevent unreliable results

>* Check types, ranges, completeness, and consistency
>* Validate carefully when combining multiple data sources

>* Check data meaning, not just format
>* Give clear warnings for unusual inputs



```
#@title Python Code - Input Validation

# Validate workflow inputs before calculations.
# Industrial data often contains mistakes.
# Clear messages support fast correction.

# Define sample production records.
records = [
    {"job": "A101", "quantity": 120, "cycle_min": 2.5},
    {"job": "B202", "quantity": -50, "cycle_min": 1.8},
    {"job": "C303", "quantity": 90, "cycle_min": 0},
]

# Prepare containers for accepted records.
valid_records = []
error_messages = []

# Validate one production record.
def validate_record(record, row_number):
    errors = []
    required_keys = {"job", "quantity", "cycle_min"}

    # Check required fields first.
    missing_keys = required_keys - set(record.keys())
    if missing_keys:
        errors.append(f"row {row_number}: missing {sorted(missing_keys)}")

    # Stop detailed checks when fields are missing.
    if missing_keys:
        return errors

    # Check job identifier quality.
    if not isinstance(record["job"], str):
        errors.append(f"row {row_number}: job must be text")

    # Check quantity type and range.
    if not isinstance(record["quantity"], int):
        errors.append(f"row {row_number}: quantity must be integer")
    elif record["quantity"] < 0:
        errors.append(f"row {row_number}: quantity cannot be negative")

    # Check cycle time type and range.
    if not isinstance(record["cycle_min"], (int, float)):
        errors.append(f"row {row_number}: cycle_min must be numeric")
    elif record["cycle_min"] <= 0:
        errors.append(f"row {row_number}: cycle_min must exceed zero")

    return errors

# Apply validation before workflow calculations.
for index, record in enumerate(records, start=1):
    record_errors = validate_record(record, index)
    if record_errors:
        error_messages.extend(record_errors)
    else:
        valid_records.append(record)

# Calculate workload only for valid records.
total_minutes = sum(
    item["quantity"] * item["cycle_min"] for item in valid_records
)

# Report compact validation results.
print(f"Valid records accepted: {len(valid_records)} of {len(records)}")
print(f"Validation errors found: {len(error_messages)}")

# Show each actionable validation message.
for message in error_messages:
    print(message)

print(f"Validated workload minutes: {total_minutes:.1f}")

```

### **3.3. Result Review**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_A/image_03_03.jpg?v=1780944472" width="250">



>* Check outputs against real-world expectations
>* Flag unusual results before decisions

>* Check outputs against real-world limits
>* Automate reviews using ranges and comparisons

>* Record checks, approvals, and flagged results
>* Use clear messages to build output confidence



```
#@title Python Code - Result Review

# Review results before trusting automated workflow outputs.
# Capacity checks compare calculations with practical limits.
# Clear flags help engineers investigate suspicious results.

# Store small production records for one work center.
records = [
    {"shift": "A", "hours": 8, "units": 960},
    {"shift": "B", "hours": 8, "units": 520},
    {"shift": "C", "hours": 7, "units": -40},
]

# Define practical limits from engineering knowledge.
max_units_per_hour = 100
min_units_per_hour = 0
expected_total_units = 1440

# Review each calculated productivity result.
flags = []
for item in records:
    rate = item["units"] / item["hours"]
    item["rate"] = rate

    # Flag impossible or unrealistic calculated rates.
    if rate < min_units_per_hour:
        flags.append(f"Shift {item['shift']}: negative production rate.")
    elif rate > max_units_per_hour:
        flags.append(f"Shift {item['shift']}: rate exceeds practical limit.")

# Compare summarized output with expected control total.
actual_total_units = sum(item["units"] for item in records)
total_difference = actual_total_units - expected_total_units
if abs(total_difference) > 50:
    flags.append(f"Total units differ by {total_difference} from plan.")

# Print a concise review summary for decision makers.
print("Result review for automated capacity report")
print(f"Reviewed shifts: {len(records)}")
print(f"Calculated total units: {actual_total_units}")
print(f"Expected total units: {expected_total_units}")

# Show only actionable review messages.
if flags:
    print("Flags requiring review:")
    for message in flags:
        print("-", message)
else:
    print("No result review flags found.")

```
