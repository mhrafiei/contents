### **1.1. Problem Selection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_B/image_01_01.jpg?v=1780945797" width="250">

>* Choose repeatable tasks with clear data flow
>* Keep projects useful, realistic, and simple

>* Help users decide, save time, reduce errors
>* Choose simple, testable, real industrial problems



```
#@title Python Code - Problem Selection

# Select practical problems for automation projects.
# Compare candidate tasks using simple criteria.
# Keep the scope realistic and testable.

# Store candidate industrial engineering problems.
problems = [
    {
        "name": "Summarize shift production notes",
        "input": "text shift notes",
        "output": "daily summary report",
        "repeatable": True,
        "user_value": 4,
        "scope_risk": 1,
    },
    {
        "name": "Predict factory demand with sensors",
        "input": "live sensor network",
        "output": "forecast dashboard",
        "repeatable": True,
        "user_value": 5,
        "scope_risk": 5,
    },
    {
        "name": "Flag overdue maintenance work orders",
        "input": "small work order list",
        "output": "overdue priority list",
        "repeatable": True,
        "user_value": 5,
        "scope_risk": 2,
    },
]

# Score problems using beginner-friendly selection logic.
def score_problem(candidate):
    clear_flow = bool(candidate["input"] and candidate["output"])
    repeat_bonus = 2 if candidate["repeatable"] else 0
    value_score = candidate["user_value"] * 2
    risk_penalty = candidate["scope_risk"] * 2
    return int(clear_flow) + repeat_bonus + value_score - risk_penalty

# Add a score to each candidate.
for candidate in problems:
    candidate["score"] = score_problem(candidate)

# Choose the highest scoring practical problem.
best_problem = max(problems, key=lambda item: item["score"])

# Print a compact planning summary.
print("Problem selection screening")
print("Candidates reviewed:", len(problems))
print("Recommended project:", best_problem["name"])
print("Input:", best_problem["input"])
print("Output:", best_problem["output"])
print("Selection score:", best_problem["score"])
print("Reason: clear data flow, repeated task, useful decision support.")

```

```
#@title Python Code - Define Project Scope

# Scope turns broad problems into manageable builds.
# This example frames a small warehouse project.
# It separates included work from excluded work.

# Create tiny order records for project planning.
orders = [
    {"area": "A", "minutes": 34, "target": 30},
    {"area": "B", "minutes": 22, "target": 25},

    {"area": "A", "minutes": 41, "target": 30},
    {"area": "C", "minutes": 28, "target": 35},
]

# Define the intended project scope clearly.
scope = {
    "process": "daily warehouse picking time review",
    "users": "shift supervisor and IE analyst",

    "input": "four sample order picking records",
    "output": "average time and late order count",
}

# Define boundaries that prevent scope creep.
out_of_scope = [
    "labor scheduling",
    "route optimization",
    "inventory accuracy",

    "seasonal demand forecasting",
]

# Verify that the tiny dataset is usable.
record_count = len(orders)
assert record_count > 0, "At least one order is required."

# Calculate simple decision-support measures.
total_minutes = sum(item["minutes"] for item in orders)
average_minutes = total_minutes / record_count
late_orders = [item for item in orders if item["minutes"] > item["target"]]

# Count late orders by warehouse area.
late_by_area = {}
for item in late_orders:
    late_by_area[item["area"]] = late_by_area.get(item["area"], 0) + 1

# Prepare a concise project planning summary.
print("Project scope:", scope["process"])
print("Primary users:", scope["users"])
print("Input data:", scope["input"])
print("Average picking time:", round(average_minutes, 1), "minutes")

# Communicate results and deliberate exclusions.
print("Late orders found:", len(late_orders), "of", record_count)
print("Late orders by area:", late_by_area)
print("Not included:", ", ".join(out_of_scope[:2]))
print("Recommended next step: verify with one full week.")

```

## **2. Build and Verify**

### **2.1. Code Implementation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_B/image_02_01.jpg?v=1780945790" width="250">



>* Turn plans into dependable Python programs
>* Use clear structure for traceable results

>* Use core Python tools thoughtfully
>* Organize functions for testing and explanation

>* Handle messy data with clear errors
>* Make assumptions visible for reliable testing



```
#@title Python Code - Code Implementation

# Build a small verification ready automation project.
# Use only built in Python features.
# Summarize production records and flag delays.

# Store tiny sample records as plain text.
raw_data = """job_id,planned_minutes,actual_minutes,good_units
A100,120,132,48
A101,90,84,39
A102,110,125,42
A103,75,70,31
"""

# Write the sample file safely.
file_path = "production_sample.csv"
with open(file_path, "w", encoding="utf-8") as file:
    file.write(raw_data)

# Convert one row into validated numbers.
def parse_record(row_text):
    parts = row_text.strip().split(",")
    if len(parts) != 4:
        raise ValueError("Expected four comma separated fields")

    job_id = parts[0]
    planned = float(parts[1])
    actual = float(parts[2])
    good_units = int(parts[3])
    return {"job": job_id, "planned": planned, "actual": actual, "units": good_units}

# Read records while skipping the header row.
def read_records(path):
    records = []
    with open(path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    for row in lines[1:]:
        records.append(parse_record(row))
    return records

# Calculate project metrics for supervisor review.
def summarize(records):
    if not records:
        raise ValueError("At least one record is required")

    total_planned = sum(item["planned"] for item in records)
    total_actual = sum(item["actual"] for item in records)
    total_units = sum(item["units"] for item in records)
    delayed = [item["job"] for item in records if item["actual"] > item["planned"]]

    throughput = total_units / (total_actual / 60)
    delay_minutes = total_actual - total_planned
    return throughput, delay_minutes, delayed

# Run the automation and verify expected results.
try:
    records = read_records(file_path)
    throughput, delay, delayed_jobs = summarize(records)
    assert len(records) == 4
    assert round(delay, 1) == 16.0

    print("Capstone build verification passed.")
    print(f"Records processed: {len(records)} jobs.")
    print(f"Total delay: {delay:.1f} minutes.")
    print(f"Throughput: {throughput:.1f} good units per hour.")
    print(f"Delayed jobs: {', '.join(delayed_jobs)}.")
except (FileNotFoundError, ValueError, AssertionError) as error:
    print(f"Verification failed: {error}")

```

### **2.2. Test the Solution**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_B/image_02_02.jpg?v=1780945792" width="250">



>* Test simple cases with known answers
>* Compare inputs, logic, and expected outputs

>* Test messy data and edge cases
>* Compare results using simple built-in checks

>* Record tests, expectations, and actual results
>* Note limitations to build trustworthy solutions



```
#@title Python Code - Test the Solution

# Build tests for a small inventory calculator.
# Compare expected answers against program outputs.
# Include normal, edge, and realistic cases.

# Define the function being verified.
def calculate_order_quantity(demand, stock, safety):
    target_stock = demand + safety
    shortage = target_stock - stock
    return max(0, shortage)

# Store small hand-checkable test cases.
test_cases = [
    (100, 20, 10, 90),
    (50, 80, 10, 0),
    (0, 0, 5, 5),

    (200, 150, 0, 50),
    (1000, 200, 100, 900),
]

# Run each test and compare values.
passed_count = 0
for case_number, case_values in enumerate(test_cases, start=1):
    demand, stock, safety, expected = case_values
    actual = calculate_order_quantity(demand, stock, safety)

    assert actual >= 0, "Order quantity became negative"
    passed = actual == expected
    passed_count += int(passed)
    print(f"Test {case_number}: expected {expected}, got {actual}, pass={passed}")

# Summarize verification evidence clearly.
total_tests = len(test_cases)
print(f"Passed {passed_count} of {total_tests} inventory tests.")

# State an important project limitation.
print("Limitation: inputs are assumed numeric and already validated.")

# Confirm the testing step finished.
print("Verification complete for the capstone build example.")

```

```
#@title Python Code - Review Results

# Review compares outputs with engineering expectations.
# Small checks reveal questionable automation results.
# Built-in Python supports practical verification.

# Create sample workstation observations.
cycle_times = [2.1, 2.0, 2.2, 1.9, 2.1]
planned_minutes = 480
allowance_rate = 0.15

# Validate important input assumptions.
record_count = len(cycle_times)
valid_times = [time for time in cycle_times if time > 0]
valid_count = len(valid_times)

# Calculate capacity using verified sample data.
average_cycle = sum(valid_times) / valid_count
available_minutes = planned_minutes * (1 - allowance_rate)
estimated_units = int(available_minutes / average_cycle)

# Compare against a manual review benchmark.
historical_best = 190
review_gap = estimated_units - historical_best
status = "reasonable" if review_gap <= 25 else "needs review"

# Summarize assumptions for decision makers.
assumptions = "All jobs available, stable staffing, no breakdowns."
limitation = "Historical comparison uses one simple benchmark."
next_step = "Test with one week of actual production data."

# Print concise verification results.
print("Review Results")
print(f"Valid observations: {valid_count} of {record_count}")
print(f"Average cycle time: {average_cycle:.2f} minutes per unit")
print(f"Estimated daily capacity: {estimated_units} units")
print(f"Historical best comparison: {review_gap:+d} units")
print(f"Engineering judgment: {status}")
print(f"Assumptions: {assumptions}")
print(f"Limitation: {limitation}")
print(f"Recommended next step: {next_step}")

```

```
#@title Python Code - Assumptions and Findings

# This script links assumptions to findings.
# Small data represents one production line.
# Printed results stay concise and useful.

# No additional package installation is required.

# Store assumptions near calculated project findings.
assumptions = {
    "scope": "One eight-hour shift",
    "breaks": "Two fifteen-minute breaks included",
    "data": "Six recent manual observations",
}

# Use tiny cycle-time observations in seconds.
cycle_times = [58, 62, 61, 59, 65, 60]

# Validate observations before using them.
valid_count = len(cycle_times) > 0
all_positive = all(time > 0 for time in cycle_times)

# Compute capacity only when data is valid.
if valid_count and all_positive:
    average_cycle = sum(cycle_times) / len(cycle_times)
    available_seconds = (8 * 60 - 30) * 60
    estimated_output = int(available_seconds // average_cycle)
else:
    average_cycle = 0
    estimated_output = 0

# Convert raw numbers into project language.
finding = "Estimated output is " + str(estimated_output) + " units per shift."
risk_note = "Finding depends on stable staffing and similar demand."

# Print assumptions and findings together.
print("ASSUMPTIONS")
print("Scope:", assumptions["scope"])
print("Breaks:", assumptions["breaks"])
print("Data:", assumptions["data"])

# Print interpreted results, not raw lists.
print("FINDINGS")
print("Average cycle time:", round(average_cycle, 1), "seconds")
print(finding)
print(risk_note)

```

### **3.2. Assumptions and Limitations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_B/image_03_02.jpg?v=1780945802" width="250">



>* State assumptions that define project trust
>* Help stakeholders judge fit and validation

>* Limitations show where results may fail
>* Clear limits prevent overclaiming project value

>* Explain how limits affect decisions
>* Show strengths, cautions, and scaling needs



```
#@title Python Code - Assumptions and Limitations

# Communicate assumptions before trusting automation outputs.
# Link limitations to operational decision risks.
# Keep capstone reporting concise and transparent.

# Create tiny capstone-style project results.
project_results = {
    "tool": "Utilization checker",
    "sample_records": 6,
    "estimated_utilization": 0.82,
}

# List assumptions supporting the calculation.
assumptions = [
    "Start and stop times are recorded accurately.",
    "Downtime reasons use consistent categories.",
    "Each machine runs only one job at once.",
]

# List limitations affecting interpretation.
limitations = [
    "No operator skill differences are modeled.",
    "Unexpected maintenance events are not predicted.",
    "Only six sample records were reviewed.",
]

# Match each limitation with a recommended next step.
next_steps = [
    "Validate with one full production week.",
    "Add maintenance calendar data later.",
    "Review categories with supervisors.",
]

# Verify the communication lists align.
list_lengths = [
    len(assumptions),
    len(limitations),
    len(next_steps),
]

# Stop if reporting sections are inconsistent.
if len(set(list_lengths)) != 1:
    raise ValueError("Reporting lists must have equal lengths.")

# Print a concise stakeholder summary.
print("CAPSTONE RESULT COMMUNICATION")
print("Tool:", project_results["tool"])
print("Records reviewed:", project_results["sample_records"])
print("Estimated utilization:", project_results["estimated_utilization"])

# Print assumptions in one compact line.
print("Key assumptions:", " | ".join(assumptions))

# Print limitations with paired actions.
for number, pair in enumerate(zip(limitations, next_steps), start=1):
    limitation, step = pair
    print(f"Limit {number}: {limitation} Next: {step}")

```

### **3.3. Recommended Next Steps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Industrial Engineers/Module_05/Lecture_B/image_03_03.jpg?v=1780945804" width="250">



>* Turn project results into practical action
>* Recommend realistic, evidence-based next steps

>* Separate quick fixes from future improvements
>* Make next steps specific and realistic

>* Identify owners, risks, and key decisions
>* Show evidence needed for responsible adoption



```
#@title Python Code - Recommended Next Steps

# Communicate next steps after project verification.
# Prioritize actions using evidence and constraints.
# Keep recommendations specific, owned, and realistic.

# Define completed capstone project evidence.
project_result = "Downtime summary tool verified on 30 maintenance logs"
manual_minutes_saved = 95
error_rate_percent = 3.3
records_tested = 30

# Capture recommended actions with owners and evidence.
next_steps = [
    ("Verify with 120 historical records", "IE analyst", 3, 5),
    ("Add missing-code warnings", "Python developer", 5, 4),
    ("Train supervisors on weekly use", "Production lead", 4, 3),
    ("Connect to maintenance database", "IT team", 2, 2),
]

# Score each action by impact and feasibility.
ranked_steps = []
for action, owner, impact, feasibility in next_steps:
    priority_score = impact + feasibility
    ranked_steps.append((priority_score, action, owner))

# Sort highest priority recommendations first.
ranked_steps.sort(reverse=True)

# Print a concise stakeholder communication summary.
print("CAPSTONE RESULT COMMUNICATION")
print("Project:", project_result)
print("Evidence:", manual_minutes_saved, "minutes saved;", error_rate_percent, "% errors.")
print("Test size:", records_tested, "records reviewed before recommendation.")
print("Recommended next steps, ranked:")

# Print ranked actions with clear ownership.
for rank, item in enumerate(ranked_steps, start=1):
    score, action, owner = item
    print(rank, action, "| owner:", owner, "| score:", score)

# State assumptions and limitations transparently.
print("Assumption: log format remains consistent next month.")
print("Limitation: peak-shift conditions were not fully tested.")
print("Decision: approve pilot after larger verification sample.")

```
