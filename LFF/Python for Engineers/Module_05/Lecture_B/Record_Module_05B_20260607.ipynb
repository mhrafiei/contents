```
#@title Python Code - Problem Definition

# Define engineering problems before coding.
# Capture scope assumptions constraints clearly.
# Turn vague requests into requirements.

# Store a broad stakeholder request.
request_text = "Check thermal chamber qualification data"

# Define candidate problem statements.
statements = [
    "Plot temperature readings from chamber sensor log.",
    "Decide whether chamber stays within plus minus two Celsius."
]

# Score statements using requirement keywords.
keywords = ["decide", "within", "sensor", "qualification"]

# Count useful requirement clues.
scores = []
for statement in statements:
    score = sum(word in statement.lower() for word in keywords)
    scores.append(score)

# Select the most actionable definition.
best_index = scores.index(max(scores))
best_statement = statements[best_index]

# Build a compact workflow contract.
contract = {
    "purpose": "support pass fail qualification decision",
    "system": "thermal chamber during acceptance test",
    "scope": "analyze one sensor file and stability window",
    "assumption": "sensor calibration is already valid",
    "success": "all readings remain within tolerance"
}

# Print a concise teaching summary.
print("Original request:", request_text)
print("Chosen definition:", best_statement)
print("Purpose:", contract["purpose"])
print("Scope:", contract["scope"])
print("Assumption:", contract["assumption"])
print("Success criterion:", contract["success"])

```

### **1.2. Approach Definition**

```
#@title Python Code - Approach Definition

# Plan engineering workflows before coding.
# Define assumptions, data, and decisions.
# Keep outputs concise for reports.

import json
import math
from pathlib import Path

# Define a tiny capstone planning dataset.
project = {
    "problem": "Estimate campus building cooling demand",
    "unit_system": "Imperial, using feet and tons",
    "method": "Simplified heat-gain calculation"
}

# Store assumptions separately for traceability.
assumptions = {
    "area_ft2": 24000,
    "peak_gain_btu_per_ft2": 18,
    "safety_factor": 1.15
}

# List workflow steps before implementation.
workflow_steps = [
    "import weather and building data",
    "validate missing or unrealistic values",
    "calculate peak cooling load",
    "compare load with equipment capacity",
    "summarize assumptions for final report"
]

# Define one reusable engineering calculation.
def estimate_cooling_tons(area_ft2, gain_rate, factor):
    load_btu_per_hr = area_ft2 * gain_rate * factor
    tons_cooling = load_btu_per_hr / 12000
    return round(tons_cooling, 2)

# Validate required numeric assumptions first.
values = list(assumptions.values())
all_positive = all(value > 0 for value in values)

# Stop safely if inputs are invalid.
if not all_positive:
    raise ValueError("All numeric assumptions must be positive.")

# Calculate one defensible preliminary result.
required_tons = estimate_cooling_tons(
    assumptions["area_ft2"],
    assumptions["peak_gain_btu_per_ft2"],
    assumptions["safety_factor"]
)

# Organize report-ready outputs clearly.
summary = {
    "engineering_question": project["problem"],
    "chosen_method": project["method"],
    "planned_steps": len(workflow_steps),
    "estimated_cooling_tons": required_tons
}

# Save a small reproducible planning file.
report_path = Path("approach_definition_summary.json")
report_path.write_text(json.dumps(summary, indent=2))

# Print concise teaching output only.
print("Approach definition example")
print(f"Question: {summary['engineering_question']}")
print(f"Method: {summary['chosen_method']}")
print(f"Workflow steps planned: {summary['planned_steps']}")
print(f"Estimated cooling capacity: {required_tons} tons")
print(f"Saved summary file: {report_path.name}")

```

### **2.1. Reusable Program Components**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_02_01.jpg?v=1780866068" width="250">



>* Build workflows from clear reusable parts

```
#@title Python Code - Reusable Program Components

# Build reusable components for engineering workflows.
# Each function has one clear responsibility.
# Small files make integration easy to test.

# Import standard tools for file handling.
import csv, tempfile, math

# Define a tiny engineering sensor dataset.
raw_rows = [
    ["time_s", "load_N", "deflection_mm"],
    ["0", "0", "0.00"],

    ["1", "120", "0.31"],
    ["2", "240", "0.63"],
    ["3", "360", "0.94"],

    ["4", "480", "bad"],
    ["5", "600", "1.57"],
]

# Write rows into a temporary CSV file.
temp_file = tempfile.NamedTemporaryFile(
    mode="w", newline="", delete=False
)

# Store the temporary path for reuse.
csv_path = temp_file.name
writer = csv.writer(temp_file)
writer.writerows(raw_rows)
temp_file.close()

# Read measurement records from a CSV file.
def read_measurements(file_path):
    with open(file_path, newline="") as handle:
        reader = csv.DictReader(handle)
        return list(reader)

# Convert valid numeric rows into dictionaries.
def clean_measurements(records):
    clean = []
    for row in records:
        try:
            item = {
                "time_s": float(row["time_s"]),
                "load_N": float(row["load_N"]),
                "deflection_mm": float(row["deflection_mm"]),
            }
        except ValueError:
            continue
        clean.append(item)
    return clean

# Estimate stiffness using least squares through origin.
def estimate_stiffness(records):
    numerator = sum(r["load_N"] * r["deflection_mm"] for r in records)
    denominator = sum(r["deflection_mm"] ** 2 for r in records)
    if denominator <= 0 or len(records) < 2:
        raise ValueError("Need multiple nonzero deflection values.")
    return numerator / denominator

# Summarize final engineering results for reporting.
def build_report(records, stiffness):
    maximum_load = max(r["load_N"] for r in records)
    maximum_deflection = max(r["deflection_mm"] for r in records)
    safe = maximum_deflection <= 2.0
    return {
        "valid_points": len(records),
        "max_load_N": maximum_load,
        "max_deflection_mm": maximum_deflection,
        "stiffness_N_per_mm": stiffness,
        "passes_limit": safe,
    }

# Run the complete workflow using reusable parts.
records = read_measurements(csv_path)
clean_records = clean_measurements(records)
stiffness_value = estimate_stiffness(clean_records)
report = build_report(clean_records, stiffness_value)

# Print a compact report for learners.
print("Reusable component workflow report")
print(f"Valid measurements: {report['valid_points']}")
print(f"Estimated stiffness: {report['stiffness_N_per_mm']:.1f} N/mm")
print(f"Maximum deflection: {report['max_deflection_mm']:.2f} mm")
print(f"Passes 2 mm limit: {report['passes_limit']}")

```

### **2.2. Function File Integration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_02_02.jpg?v=1780866066" width="250">



>* Connect small functions into one workflow
>* Use consistent inputs, outputs, and names

>* Define each function’s role clearly
>* Pass data consistently and test integration

>* Connected functions support teamwork and updates
>* Repeatable workflows produce trustworthy engineering results



```
#@title Python Code - Function File Integration

# This script integrates reusable engineering workflow functions.
# Small files simulate capstone project data exchange.
# Results stay compact for classroom readability.

import csv
from pathlib import Path

# Store temporary data in a small file.
data_path = Path("capstone_flow_measurements.csv")
report_path = Path("capstone_report.txt")

# Create reproducible measurement records.
rows = [
    ["station", "flow_gpm", "diameter_in"],
    ["A", "95", "4.0"],

    ["B", "110", "4.0"],
    ["C", "", "4.0"],
    ["D", "125", "4.0"],
]

# Write the raw engineering measurements.
with data_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Read records from a reusable file.
def load_measurements(path):
    with path.open("r", newline="") as file:
        reader = csv.DictReader(file)
        return list(reader)

# Convert missing entries using the mean.
def clean_flow_values(records):
    known = [float(row["flow_gpm"]) for row in records if row["flow_gpm"]]
    average = sum(known) / len(known)

    cleaned = []
    for row in records:
        value = float(row["flow_gpm"] or average)
        cleaned.append((row["station"], value, float(row["diameter_in"])))
    return cleaned

# Convert gallons per minute into cubic meters per second.
def convert_units(cleaned):
    converted = []
    for station, flow_gpm, diameter_in in cleaned:
        flow_m3s = flow_gpm * 0.0000630902
        diameter_m = diameter_in * 0.0254
        converted.append((station, flow_m3s, diameter_m))
    return converted

# Apply a numerical hydraulic velocity calculation.
def compute_velocity(converted):
    results = []
    for station, flow_m3s, diameter_m in converted:
        area = 3.14159 * (diameter_m / 2) ** 2
        velocity = flow_m3s / area
        results.append((station, velocity))
    return results

# Summarize outputs for a report file.
def summarize_results(results):
    speeds = [speed for station, speed in results]
    return {
        "count": len(speeds),
        "average_velocity": sum(speeds) / len(speeds),
        "maximum_velocity": max(speeds),
    }

# Save a compact reproducible report.
def save_report(path, summary):
    lines = [
        "Capstone hydraulic workflow report",
        f"Stations analyzed: {summary['count']}",
        f"Average velocity: {summary['average_velocity']:.2f} m/s",
        f"Maximum velocity: {summary['maximum_velocity']:.2f} m/s",
    ]

    path.write_text("\n".join(lines))
    return lines

# Run the integrated function workflow.
records = load_measurements(data_path)
cleaned = clean_flow_values(records)
converted = convert_units(cleaned)
velocities = compute_velocity(converted)
summary = summarize_results(velocities)
report_lines = save_report(report_path, summary)

# Check correctness before displaying results.
assert summary["count"] == len(records)
assert summary["average_velocity"] > 0

# Print concise integrated workflow evidence.
print("Integrated function files completed successfully.")
print(f"Input file: {data_path.name}")
print(f"Output file: {report_path.name}")
print(report_lines[2])
print(report_lines[3])

```

### **2.3. Coherent Program Assembly**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_02_03.jpg?v=1780866070" width="250">



>* Combine project parts into one workflow

>* Assign clear roles to program components

>* Build programs others can understand and trust
>* Organize workflows for reuse and verification



```
#@title Python Code - Coherent Program Assembly

# Assemble one small engineering workflow.
# Keep each program responsibility visible.
# Use files, functions, and checks.

import csv
from pathlib import Path

# Store project files together.
project_dir = Path("capstone_demo_outputs")
project_dir.mkdir(exist_ok=True)

# Define a tiny rainfall dataset.
rain_file = project_dir / "rainfall_inches.csv"
rows = [(0, 0.00), (1, 0.35), (2, 0.80), (3, 0.25)]

# Write reproducible input data.
with rain_file.open("w", newline="") as file_obj:
    writer = csv.writer(file_obj)
    writer.writerow(["hour", "rain_in"])

    writer.writerows(rows)

# Read rainfall records from file.
def read_rainfall(path):
    with path.open(newline="") as file_obj:
        reader = csv.DictReader(file_obj)
        data = [(int(row["hour"]), float(row["rain_in"])) for row in reader]

    return data

# Validate before engineering calculations.
def validate_records(records):
    if len(records) < 2:
        raise ValueError("Need at least two time records")

    hours = [item[0] for item in records]
    rain = [item[1] for item in records]

    if hours != sorted(hours):
        raise ValueError("Hours must be increasing")

    if any(value < 0 for value in rain):
        raise ValueError("Rainfall cannot be negative")

# Compute runoff using simple assumptions.
def compute_runoff(records, area_acres, runoff_coeff):
    total_rain = sum(item[1] for item in records)
    runoff_in = total_rain * runoff_coeff

    volume_ft3 = runoff_in / 12 * area_acres * 43560
    return {"rain_in": total_rain, "runoff_ft3": volume_ft3}

# Export concise report information.
def write_summary(path, summary, limit_ft3):
    status = "OK" if summary["runoff_ft3"] <= limit_ft3 else "REVIEW"
    with path.open("w", newline="") as file_obj:

        writer = csv.writer(file_obj)
        writer.writerow(["metric", "value"])
        writer.writerow(["total_rain_in", round(summary["rain_in"], 2)])
        writer.writerow(["runoff_ft3", round(summary["runoff_ft3"], 1)])

        writer.writerow(["design_status", status])
    return status

# Assemble the full capstone workflow.
records = read_rainfall(rain_file)
validate_records(records)
summary = compute_runoff(records, area_acres=2.0, runoff_coeff=0.65)

# Produce organized program output.
summary_file = project_dir / "runoff_summary.csv"
status = write_summary(summary_file, summary, limit_ft3=8000)

# Print only decision focused results.
print("Workflow completed from input file to summary file.")
print(f"Records checked: {len(records)} rainfall observations.")
print(f"Total rainfall: {summary['rain_in']:.2f} inches.")
print(f"Estimated runoff: {summary['runoff_ft3']:.1f} cubic feet.")
print(f"Design status: {status}; summary saved at {summary_file}.")

```

### **3.1. Validation Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_03_01.jpg?v=1780866077" width="250">



>* Check data before trusting program results
>* Flag suspicious inputs using engineering judgment

>* Check workflow logic between program stages
>* Verify units, fields, and unexpected data

>* Clear messages help future users inspect workflows
>* Documented assumptions support defensible, reproducible results



```
#@title Python Code - Validation Checks

# Validation checks protect engineering workflow decisions.
# Small examples reveal common data problems.
# Clear messages improve reproducible program reviews.

# Use built-in Python only here.
import math

# Define expected engineering input limits.
required_fields = ["station", "length_m", "load_kN", "temperature_K"]

# Create small capstone-style input records.
records = [
    {"station": "A", "length_m": 3.2, "load_kN": 18.5, "temperature_K": 294.0},
    {"station": "B", "length_m": 0.0, "load_kN": 12.1, "temperature_K": 289.0},
    {"station": "C", "length_m": 4.1, "load_kN": -2.0, "temperature_K": 301.0},
    {"station": "D", "length_m": 2.7, "load_kN": 15.0, "temperature_K": -5.0},
]

# Build reusable validation logic.
def validate_record(record, index):
    issues = []

    # Check for missing required fields.
    for field in required_fields:
        if field not in record:
            issues.append(f"row {index}: missing {field}")

    # Stop deeper checks after missing data.
    if issues:
        return issues

    # Check numeric values before calculations.
    numeric_fields = ["length_m", "load_kN", "temperature_K"]
    for field in numeric_fields:
        value = record[field]

        if not isinstance(value, (int, float)):
            issues.append(f"row {index}: {field} is not numeric")
        elif not math.isfinite(value):
            issues.append(f"row {index}: {field} is not finite")

    # Check engineering assumptions and physical limits.
    if record["length_m"] <= 0:
        issues.append(f"row {index}: length must be positive")

    if record["load_kN"] < 0:
        issues.append(f"row {index}: load cannot be negative")

    if record["temperature_K"] <= 0:
        issues.append(f"row {index}: absolute temperature invalid")

    return issues

# Run validation before engineering calculations.
all_issues = []
valid_records = []

for index, record in enumerate(records, start=1):
    issues = validate_record(record, index)

    if issues:
        all_issues.extend(issues)
    else:
        valid_records.append(record)

# Calculate only with validated records.
average_load = sum(item["load_kN"] for item in valid_records)
average_load = average_load / max(len(valid_records), 1)

# Report concise review results.
print(f"Records reviewed: {len(records)}")
print(f"Valid records used: {len(valid_records)}")
print(f"Validation issues found: {len(all_issues)}")

# Show limited messages for user action.
for message in all_issues[:5]:
    print(message)

# Summarize safe calculation outcome.
print(f"Average valid load: {average_load:.2f} kN")

```

### **3.2. Correctness Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_03_02.jpg?v=1780866079" width="250">



>* Check results match engineering principles
>* Compare outputs with trusted known answers

>* Test functions and workflows at multiple scales
>* Check realistic scenarios and boundary cases

>* Trace data through every workflow step
>* Check evidence before trusting final results



```
#@title Python Code - Correctness Checks

# Correctness checks compare code with trusted engineering expectations.
# This example tests beam deflection calculations systematically.
# Small cases reveal mistakes before final reporting.

import math

# Define material and beam values.
load_newtons = 1000.0
length_meters = 2.0
modulus_pascals = 200_000_000_000.0
inertia_meters4 = 8.0e-6

# Compute cantilever tip deflection.
def beam_deflection(load, length, modulus, inertia):
    numerator = load * length ** 3
    denominator = 3.0 * modulus * inertia

    return numerator / denominator

# Store expected hand calculation result.
expected_meters = 0.0016666666666666668
computed_meters = beam_deflection(load_newtons, length_meters, modulus_pascals, inertia_meters4)

# Check agreement with a strict tolerance.
tolerance_meters = 1.0e-12
passes_known_answer = math.isclose(computed_meters, expected_meters, abs_tol=tolerance_meters)

# Check physical behavior for doubled load.
doubled_load_value = beam_deflection(2.0 * load_newtons, length_meters, modulus_pascals, inertia_meters4)
passes_scaling = math.isclose(doubled_load_value, 2.0 * computed_meters, rel_tol=1.0e-12)

# Check boundary behavior for zero load.
zero_load_value = beam_deflection(0.0, length_meters, modulus_pascals, inertia_meters4)
passes_boundary = math.isclose(zero_load_value, 0.0, abs_tol=tolerance_meters)

# Summarize correctness evidence compactly.
all_checks_pass = passes_known_answer and passes_scaling and passes_boundary
print(f"Known-answer check passed: {passes_known_answer}")
print(f"Load-scaling check passed: {passes_scaling}")
print(f"Zero-load boundary check passed: {passes_boundary}")
print(f"Computed deflection: {computed_meters * 1000.0:.3f} mm")
print(f"Overall correctness review passed: {all_checks_pass}")

```

### **3.3. Reproducibility Review**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_05/Lecture_B/image_03_03.jpg?v=1780866081" width="250">



>* Recreate results under documented conditions
>* Provide files, assumptions, versions, and instructions

>* Control small choices that change results
>* Trace inputs to verified final outputs

>* Document enough for future engineers
>* Ensure workflows can be rerun and extended



```
#@title Python Code - Reproducibility Review

# Reproducibility review checks rerunnable engineering workflows.
# This example records assumptions and versions.
# Deterministic outputs make audits easier later.

import csv
import json
import random
import platform

from pathlib import Path
from statistics import mean
from datetime import datetime, timezone

# Store reproducible inputs inside one project folder.
project_folder = Path("reproducible_capstone_demo")
project_folder.mkdir(exist_ok=True)
input_file = project_folder / "bridge_load_inputs.csv"

# Fixed rows replace hidden manual spreadsheet edits.
rows = [
    {"span_id": "A", "load_kips": 42.0},
    {"span_id": "B", "load_kips": 39.5},
    {"span_id": "C", "load_kips": 44.2},
]

# Write the tiny input file deterministically.
with input_file.open("w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=rows[0].keys())
    writer.writeheader()
    writer.writerows(rows)

# Record assumptions beside the input data.
assumptions = {
    "unit": "kips",
    "safety_factor": 1.25,
    "random_seed": 7,
}

# Save review metadata in a readable JSON file.
metadata_file = project_folder / "run_metadata.json"
metadata = {
    "python_version": platform.python_version(),
    "created_utc": datetime.now(timezone.utc).isoformat(),
    "assumptions": assumptions,
}

with metadata_file.open("w") as file:
    json.dump(metadata, file, indent=2)

# Use a fixed seed for reproducible simulation noise.
random.seed(assumptions["random_seed"])
loads = [float(row["load_kips"]) for row in rows]
variation = [random.uniform(-0.5, 0.5) for _ in loads]

# Compute a final result from traceable inputs.
adjusted = [load + delta for load, delta in zip(loads, variation)]
design_load = mean(adjusted) * assumptions["safety_factor"]
output_file = project_folder / "design_load_result.txt"

# Save the computed result without manual editing.
with output_file.open("w") as file:
    file.write(f"design_load_kips={design_load:.3f}\n")

# Summarize the reproducibility review visibly.
print("Reproducibility review summary")
print(f"Project folder: {project_folder}")
print(f"Input rows recorded: {len(rows)}")
print(f"Python version recorded: {metadata['python_version']}")
print(f"Random seed recorded: {assumptions['random_seed']}")
print(f"Design load recreated: {design_load:.3f} kips")
print("Files created: inputs, metadata, result")
print("Rerun this cell to confirm matching results")

```
