### **1.1. Defining Function Purpose**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780716638" width="250">



>* Give each function one clear calculation purpose
>* Keep workflow tasks outside focused functions

>* State each function’s purpose clearly
>* Avoid combining unrelated engineering tasks

>* Tie functions to specific engineering calculations
>* Use focused functions for modular, maintainable code



```
#@title Python Code - Defining Function Purpose

# Focus each function on one engineering calculation.
# Keep computation separate from user-facing output.
# Document units and assumptions before implementation.

def required_area(load_newtons, allowable_stress_pascals):
    """Return required area in square meters."""
    if load_newtons <= 0:
        raise ValueError("load must be positive")

    if allowable_stress_pascals <= 0:
        raise ValueError("allowable stress must be positive")
    area_square_meters = load_newtons / allowable_stress_pascals
    return area_square_meters


def format_area_report(area_square_meters):
    """Return a short display message."""
    area_square_millimeters = area_square_meters * 1_000_000
    return f"Required area: {area_square_millimeters:.1f} mm^2"


# Provide clear engineering inputs with units.
applied_load_newtons = 12000
allowable_stress_pascals = 150_000_000

# Call the calculation function first.
required_area_value = required_area(
    applied_load_newtons,
    allowable_stress_pascals,
)

# Format output separately from calculation logic.
summary_message = format_area_report(required_area_value)
print("Focused function example")
print(summary_message)

# Reuse calculation without changing output code.
second_area_value = required_area(18000, allowable_stress_pascals)
print(format_area_report(second_area_value))

```

### **1.2. Parameter Planning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780716640" width="250">



>* Choose inputs needed for focused calculations
>* Make data clear, reusable, and checkable

>* Use meaningful parameter names and units
>* Include only inputs the calculation needs

>* Use defaults carefully and document assumptions
>* Keep inputs explicit, testable, and focused



```
#@title Python Code - Parameter Planning

# Plan parameters before writing engineering functions.
# Name inputs using quantities and units.
# Return results without printing inside functions.

# Define a focused pump power calculation.
def hydraulic_power_kw(flow_m3_s, pressure_rise_pa, efficiency=1.0):
    """Return pump power in kW for SI inputs."""

    # Validate physically meaningful parameter choices.
    if flow_m3_s < 0 or pressure_rise_pa < 0:
        raise ValueError("Flow and pressure must be nonnegative.")

    # Validate efficiency as a realistic fraction.
    if not 0 < efficiency <= 1:
        raise ValueError("Efficiency must be between zero and one.")

    # Compute hydraulic power using visible parameters.
    power_watts = flow_m3_s * pressure_rise_pa / efficiency
    return power_watts / 1000

# Define example inputs with explicit units.
flow_m3_s = 0.035
pressure_rise_pa = 180000
pump_efficiency = 0.72

# Call the function with named arguments.
power_kw = hydraulic_power_kw(
    flow_m3_s=flow_m3_s,
    pressure_rise_pa=pressure_rise_pa,
    efficiency=pump_efficiency,
)

# Display only the reusable returned result.
print("Parameter planning example:")
print(f"Flow rate: {flow_m3_s} m^3/s")
print(f"Pressure rise: {pressure_rise_pa} Pa")
print(f"Pump efficiency: {pump_efficiency}")
print(f"Required shaft power: {power_kw:.2f} kW")

```

### **1.3. Function Scope**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780716644" width="250">



>* Limit each function to one calculation
>* Focused scope improves testing and reuse

>* Pass only relevant calculation data
>* Use explicit parameters for trustworthy results

>* Design functions for reuse across related problems
>* Keep responsibilities focused and side effects minimal



```
#@title Python Code - Function Scope

# This example demonstrates focused function scope.
# Each function receives clear engineering parameters.
# Calculations return values instead of printing.

# Define a focused cylindrical tank volume function.
def tank_volume(radius_m, height_m):
    """Return cylinder volume in cubic meters."""

    # Validate dimensions before calculating volume.
    if radius_m <= 0 or height_m <= 0:
        raise ValueError("dimensions must be positive")

    # Keep the calculation inside this function.
    volume_m3 = 3.141592653589793 * radius_m**2 * height_m
    return volume_m3

# Define a separate mass calculation function.
def water_mass(volume_m3, density_kg_m3=1000.0):
    """Return water mass using consistent SI units."""

    # Validate physical inputs for this calculation.
    if volume_m3 <= 0 or density_kg_m3 <= 0:
        raise ValueError("inputs must be positive")

    # Return the result without printing details.
    mass_kg = volume_m3 * density_kg_m3
    return mass_kg

# Supply explicit inputs from the workflow.
radius_m = 1.2
height_m = 3.5

# Reuse calculation functions in sequence.
volume_m3 = tank_volume(radius_m, height_m)
mass_kg = water_mass(volume_m3)

# Print only user-facing summary outside functions.
print("Focused engineering functions example")
print(f"Tank volume: {volume_m3:.2f} m^3")
print(f"Water mass: {mass_kg:.0f} kg")
print("Functions computed results without managing the whole workflow.")

```

## **2. Return Values**

### **2.1. Returning Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780716651" width="250">



>* Return values separate calculation from presentation
>* Reusable results support testing and workflows

>* Printing shows results; returning preserves values.
>* Returned values support further engineering calculations.

>* Functions should calculate and return trusted outputs
>* Presentation changes stay outside stable calculations



```
#@title Python Code - Returning Results

# Functions should return reusable engineering calculation results.
# Printing belongs outside focused calculation functions.
# Returned values support testing and later workflows.

# Define a focused calculation function.
def beam_stress(force_n, area_m2):
    """Return axial stress in pascals from force and area."""

    # Validate physical input assumptions.
    if area_m2 <= 0:
        raise ValueError("Area must be positive.")

    # Return the calculated numerical result.
    return force_n / area_m2

# Store inputs with clear engineering units.
force_n = 12000.0
area_m2 = 0.003

# Capture returned result for later reuse.
stress_pa = beam_stress(force_n, area_m2)
stress_mpa = stress_pa / 1_000_000

# Reuse the returned value in a decision.
allowable_mpa = 5.0
passes_check = stress_mpa <= allowable_mpa

# Present results outside the calculation function.
print(f"Stress: {stress_mpa:.2f} MPa")
print(f"Allowable: {allowable_mpa:.2f} MPa")
print(f"Design passes check: {passes_check}")

```

### **2.2. Returning Calculated Results**

>* Return values for later program use
>* Keep calculations separate from display choices

>* Returned numbers are easier to test
>* Separate calculations stay flexible and reusable

>* Return results from focused calculation functions
>* Reuse outputs while keeping responsibilities clear



```
#@title Python Code - Returning Calculated Results

# Functions should return engineering results.
# Presentation belongs outside calculation functions.
# Returned values support later workflows.

# Define a focused engineering calculation.
def tank_volume_cubic_meters(flow_m3_per_hour, storage_hours):
    """Return tank volume in cubic meters."""
    return flow_m3_per_hour * storage_hours

# Define inputs with clear engineering units.
design_flow_rate = 2.5
storage_duration = 6.0

# Store the returned numeric result.
required_volume = tank_volume_cubic_meters(
    design_flow_rate,
    storage_duration,
)

# Reuse the result in separate decisions.
selected_tank_size = 20.0
remaining_capacity = selected_tank_size - required_volume

# Present results outside the function.
print("Required tank volume:", required_volume, "m^3")
print("Selected tank size:", selected_tank_size, "m^3")
print("Remaining capacity:", remaining_capacity, "m^3")

```

### **2.3. Returning Computed Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780716653" width="250">



>* Return values for later program use
>* Keep calculation separate from presentation

>* Return values instead of printing answers
>* Handle display outside calculation functions

>* Returned values make testing clearer
>* Separated output keeps programs maintainable



```
#@title Python Code - Returning Computed Results

# Functions should return reusable engineering results.
# Output belongs outside focused calculation functions.
# Docstrings record assumptions, inputs, and units.

def wall_heat_loss(area_m2, thickness_m, conductivity_w_mk, delta_t_k):
    """Return steady heat loss in watts, using one-dimensional conduction."""

    # Validate physical inputs before calculation.
    values = [area_m2, thickness_m, conductivity_w_mk, delta_t_k]

    # Reject nonpositive dimensions and material properties.
    if area_m2 <= 0 or thickness_m <= 0 or conductivity_w_mk <= 0:
        raise ValueError("area, thickness, and conductivity must be positive")

    # Compute Fourier conduction heat transfer rate.
    heat_loss_w = conductivity_w_mk * area_m2 * delta_t_k / thickness_m

    return heat_loss_w

# Define a small wall design example.
area_m2 = 12.0
thickness_m = 0.18
conductivity_w_mk = 0.045

# Define the indoor outdoor temperature difference.
delta_t_k = 22.0
heat_loss_w = wall_heat_loss(area_m2, thickness_m, conductivity_w_mk, delta_t_k)

# Reuse the returned value in another calculation.
hours_per_year = 24 * 180
energy_kwh = heat_loss_w * hours_per_year / 1000

# Display results outside the calculation function.
print(f"Returned heat loss: {heat_loss_w:.1f} W")
print(f"Estimated seasonal energy: {energy_kwh:.1f} kWh")
print("The function returned numbers, not report text.")

```

## **3. Documenting Engineering Functions**

### **3.1. Docstring Essentials**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780716646" width="250">



>* Explain purpose, assumptions, and intended use
>* Connect engineering reasoning to code implementation

>* State units for inputs and outputs
>* Document conventions to prevent unsafe results

>* State model limits and key assumptions
>* Keep docstrings concise, useful, and trustworthy



```
#@title Python Code - Docstring Essentials

# Docstrings record engineering intent clearly.
# Units and assumptions prevent misuse.
# Return values stay separate from output.

def beam_midspan_deflection(load_newtons, span_meters, modulus_pascals, inertia_m4):
    """Return midspan deflection in meters.

    Assumes a simply supported beam with one centered point load.

    Inputs use newtons, meters, pascals, and meters to fourth power.

    Uses small deflection theory and linear elastic material behavior.
    """
    # Validate all inputs before calculation.
    inputs_are_positive = all(value > 0 for value in (
        load_newtons, span_meters, modulus_pascals, inertia_m4
    ))

    # Stop clearly when assumptions are violated.
    if not inputs_are_positive:
        raise ValueError("All engineering inputs must be positive values.")

    # Apply the documented beam formula.
    deflection_meters = load_newtons * span_meters**3
    deflection_meters /= 48 * modulus_pascals * inertia_m4

    # Return the result without printing inside.
    return deflection_meters

# Define a small steel beam example.
load_newtons = 1200.0
span_meters = 2.5
modulus_pascals = 200e9
inertia_m4 = 8.5e-6

# Call the function using documented units.
deflection_meters = beam_midspan_deflection(
    load_newtons, span_meters, modulus_pascals, inertia_m4
)

# Convert meters to millimeters for display.
deflection_millimeters = deflection_meters * 1000.0
summary_line = f"Estimated deflection: {deflection_millimeters:.3f} mm"
assumption_line = "Docstring states units, loading, supports, and model limits."

# Show concise output for the lecture.
print(summary_line)
print(assumption_line)

```

### **3.2. Code Comments**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780716648" width="250">



>* Explain engineering assumptions, not obvious code
>* Connect calculations to modeling decisions

>* Place comments where key choices matter
>* Avoid clutter from obvious or excessive comments

>* Comments flag limits when functions are reused
>* They preserve assumptions and key implementation reasoning



```
#@title Python Code - Code Comments

# Engineering comments should explain modeling choices.
# This example separates calculation from display.
# Units and assumptions stay near calculations.

def estimate_beam_stress(load_newtons, span_meters, width_mm, height_mm):
    """Return maximum bending stress in MPa for a rectangular beam."""
    if min(load_newtons, span_meters, width_mm, height_mm) <= 0:
        raise ValueError("All engineering inputs must be positive.")

    # Simply supported beam with central point load.
    max_moment_newton_meter = load_newtons * span_meters / 4
    width_meters = width_mm / 1000
    height_meters = height_mm / 1000

    # Convert millimeters because stress equations require meters.
    section_modulus_m3 = width_meters * height_meters ** 2 / 6
    stress_pascal = max_moment_newton_meter / section_modulus_m3
    stress_mpa = stress_pascal / 1_000_000

    # Returning the value keeps calculation reusable elsewhere.
    return stress_mpa

load_newtons = 1200
span_meters = 1.8
width_mm = 40
height_mm = 80

stress_mpa = estimate_beam_stress(
    load_newtons, span_meters, width_mm, height_mm
)

limit_mpa = 150
status_text = "acceptable" if stress_mpa <= limit_mpa else "too high"
print(f"Estimated stress: {stress_mpa:.1f} MPa")
print(f"Design limit: {limit_mpa} MPa")
print(f"Status: {status_text}")

```

### **3.3. Expected Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780716650" width="250">



>* State required input meaning and units
>* Reduce misuse and improve function reuse

>* Specify compatible units and valid input conditions
>* Consider edge cases and model limits

>* Concise inputs guide future function users
>* Note assumptions and non-obvious input choices



```
#@title Python Code - Expected Inputs

# Engineering functions need clear input expectations.
# Docstrings record assumptions and required units.
# Validation catches common input mistakes early.

def heat_energy_kj(mass_kg, specific_heat_kj_per_kg_k, delta_temp_k):
    """Return heat energy in kJ for positive mass and temperature change.

    Expected inputs:
    mass_kg: fluid mass in kilograms, greater than zero.
    specific_heat_kj_per_kg_k: heat capacity in kJ/(kg*K), greater than zero.
    delta_temp_k: temperature change in kelvin, zero or greater.
    Assumes no phase change and constant specific heat.
    """
    # Check mass before using it.
    if mass_kg <= 0:
        raise ValueError("mass_kg must be greater than zero")

    # Check heat capacity before calculation.
    if specific_heat_kj_per_kg_k <= 0:
        raise ValueError("specific_heat must be greater than zero")

    # Check temperature change assumption.
    if delta_temp_k < 0:
        raise ValueError("delta_temp_k must be zero or positive")

    # Return only the computed engineering result.
    return mass_kg * specific_heat_kj_per_kg_k * delta_temp_k

# Use documented units for this example.
water_mass_kg = 2.5
water_heat_capacity = 4.186
water_temperature_rise_k = 18.0

# Call the function with expected inputs.
energy_needed = heat_energy_kj(
    water_mass_kg,
    water_heat_capacity,
    water_temperature_rise_k
)

# Show concise output for learners.
print("Function:", heat_energy_kj.__name__)
print("Expected input units: kg, kJ/(kg*K), K")
print("Assumption: no phase change, constant specific heat")
print("Energy needed:", round(energy_needed, 2), "kJ")

```
