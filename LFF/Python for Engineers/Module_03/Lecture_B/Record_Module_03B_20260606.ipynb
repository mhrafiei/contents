## **1. Reading Data Files**

### **1.1. Opening Data Files**

>* Open files to access engineering measurements
>* Use read mode to protect original data

>* File names identify; paths locate files

>* Check location, permissions, format, and mode



```
#@title Python Code - Opening Data Files

# Create and open small engineering data files.
# Use safe read mode for original records.
# Close files automatically with context managers.

from pathlib import Path

# Store example data beside this notebook.
data_path = Path("temperature_log.txt")

# Write a tiny sample sensor record.
data_path.write_text("time_s,temp_C\n0,22.4\n5,23.1\n10,23.8\n")

# Confirm the file exists before opening.
if data_path.exists() and data_path.is_file():

    # Open the file in read-only text mode.
    with open(data_path, "r", encoding="utf-8") as file_object:

        # Read only the first two lines.
        header_line = file_object.readline().strip()
        first_data_line = file_object.readline().strip()

    # Show the opened file details.
    print("Opened file:", data_path.name)
    print("Mode used: read-only text")
    print("Header:", header_line)
    print("First measurement:", first_data_line)

else:

    # Report missing files without crashing.
    print("File was not found:", data_path)

print("File closed automatically:", file_object.closed)

```

### **1.2. Line Reading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1780718597" width="250">



>* Process engineering records one line at a time
>* Inspect and convert data efficiently

>* Clean and split text before calculations
>* Handle headers, units, and formats carefully

>* Check each line for data problems
>* Turn raw file text into reliable data



```
#@title Python Code - Line Reading

# Read engineering data one line at a time.
# Clean text before converting numerical values.
# Skip headers, blanks, and faulty records.

# Create a small sample sensor log.
file_name = "bridge_strain_log.csv"

# Store simple CSV-style engineering records.
sample_text = "time_s,strain_micro,temperature_C\n0,12.5,21.0\n1,13.1,21.1\n\n2,missing,21.2\n3,14.0,21.4\n"

# Write the sample data safely.
with open(file_name, "w", encoding="utf-8") as file:
    file.write(sample_text)

# Prepare containers for valid readings.
times = []
strains = []
temperatures = []
skipped_rows = 0

# Open the file for line reading.
with open(file_name, "r", encoding="utf-8") as file:
    header = file.readline().strip()

    # Process each remaining record line.
    for line_number, line in enumerate(file, start=2):
        cleaned_line = line.strip()

        # Ignore empty lines safely.
        if cleaned_line == "":
            skipped_rows += 1
            continue

        # Split comma separated values.
        parts = cleaned_line.split(",")

        # Check the expected column count.
        if len(parts) != 3:
            skipped_rows += 1
            continue

        # Convert text values defensively.
        try:
            time_s = float(parts[0])
            strain = float(parts[1])
            temp_c = float(parts[2])

        except ValueError:
            skipped_rows += 1
            continue

        # Save only valid numerical records.
        times.append(time_s)
        strains.append(strain)
        temperatures.append(temp_c)

# Check that useful data was read.
if len(strains) > 0:
    average_strain = sum(strains) / len(strains)
else:
    average_strain = None

# Print a compact reading summary.
print("Header read:", header)
print("Valid data rows:", len(strains))
print("Skipped rows:", skipped_rows)
print("First valid strain:", strains[0])
print("Average strain:", round(average_strain, 2))

```

### **1.3. CSV Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1780718600" width="250">



>* CSV files store engineering measurements as text
>* Read lines, split fields, interpret structure

>* Handle headers, units, and notes carefully
>* Split fields and convert text to numbers

>* Watch for messy or missing CSV values
>* Validate rows before engineering analysis



```
#@title Python Code - CSV Data

# Create a small engineering CSV file.
# Read rows using built-in file handling.
# Convert text fields into useful numbers.

# No extra packages are required.

# Store sample lines with one flawed row.
csv_lines = [
    "time_s,temperature_C,pressure_kPa",
    "0.0,22.5,101.3",

    "1.0,23.1,101.5",
    "2.0,,101.8",
    "3.0,24.0,102.0",
]

# Choose a small local filename.
file_name = "sensor_readings.csv"

# Write the sample CSV text file.
with open(file_name, "w", encoding="utf-8") as file:
    file.write("\n".join(csv_lines))

# Prepare containers for valid records.
valid_rows = []
skipped_rows = []

# Open the file for reading.
with open(file_name, "r", encoding="utf-8") as file:
    header = file.readline().strip().split(",")

    # Read remaining lines one by one.
    for line_number, line in enumerate(file, start=2):
        parts = line.strip().split(",")

        # Check field count before converting.
        if len(parts) != len(header):
            skipped_rows.append(line_number)
            continue

        # Convert text fields safely.
        try:
            time_s = float(parts[0])
            temp_c = float(parts[1])
            pressure_kpa = float(parts[2])

        except ValueError:
            skipped_rows.append(line_number)
            continue

        # Store one valid engineering observation.
        valid_rows.append((time_s, temp_c, pressure_kpa))

# Extract temperature values for analysis.
temperatures = [row[1] for row in valid_rows]
pressures = [row[2] for row in valid_rows]

# Compute simple summaries from valid data.
avg_temp = sum(temperatures) / len(temperatures)
avg_pressure = sum(pressures) / len(pressures)

# Print a compact report under limits.
print(f"Columns found: {header}")
print(f"Valid data rows: {len(valid_rows)}")
print(f"Skipped row numbers: {skipped_rows}")
print(f"Average temperature: {avg_temp:.2f} C")
print(f"Average pressure: {avg_pressure:.2f} kPa")

```

## **2. Writing Reports**

### **2.1. Text Report Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1780718593" width="250">



>* Turn calculations into clear readable records
>* Add labels units and structure to prevent errors

>* Start reports with context and organized results
>* Use labels, spacing, and units clearly

>* Use suitable precision without clutter
>* Keep units, styles, and layout consistent



```
#@title Python Code - Text Report Formatting

# Create a clear engineering text report.
# Format labels, units, and numbers.
# Write and preview a small file.

# Define compact engineering calculation results.
project_name = "Beam Deflection Check"
span_m = 2.4
load_n = 850.0
max_deflection_mm = 3.2764

# Choose readable report file name.
report_path = "beam_deflection_report.txt"
allowable_deflection_mm = span_m * 1000 / 360
safety_factor = allowable_deflection_mm / max_deflection_mm
status = "PASS" if safety_factor >= 1 else "FAIL"

# Build formatted lines with context.
report_lines = [
    "BEAM DEFLECTION REPORT",
    "======================",
    f"Project: {project_name}",
    f"Span: {span_m:.2f} m",
]

# Add calculation results with units.
report_lines.extend([
    f"Applied load: {load_n:.0f} N",
    f"Maximum deflection: {max_deflection_mm:.2f} mm",
    f"Allowable deflection: {allowable_deflection_mm:.2f} mm",
    f"Safety factor: {safety_factor:.2f}",
])

# Add short interpretation line.
report_lines.extend([
    "",
    f"Result: {status}",
    "Note: Values are rounded for reporting clarity.",
])

# Write the report safely.
try:
    with open(report_path, "w", encoding="utf-8") as file:
        file.write("\n".join(report_lines))
except OSError as error:
    print(f"Could not write report: {error}")

# Preview the saved report briefly.
with open(report_path, "r", encoding="utf-8") as file:
    preview_lines = file.read().splitlines()[:10]

# Print the formatted report preview.
print("Saved report preview:")
print("\n".join(preview_lines))

```

### **2.2. Formatted File Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1780718590" width="250">



>* Write results clearly with labels and units
>* Create reports others can review and reuse

>* Control number precision for meaningful reports
>* Use consistent formatting, spacing, and units

>* Format reports for readers and programs
>* Include context for clear future reuse



```
#@title Python Code - Formatted File Output

# This example writes formatted engineering reports.
# Labeled values make output easier to review.
# Defensive checks prevent invalid report calculations.

# Define a small pump test dataset.
flow_lps = [8.0, 10.0, 12.0]
pressure_kpa = [180.0, 210.0, 225.0]
power_kw = [2.10, 2.85, 3.45]

# Check list lengths before calculation.
if len(flow_lps) == len(pressure_kpa) == len(power_kw):
    row_count = len(flow_lps)
else:
    raise ValueError("Input lists must have matching lengths.")

# Use constants with clear engineering units.
water_density = 1000.0
report_path = "pump_performance_report.txt"
efficiencies = []

# Calculate efficiency for each operating point.
for flow, pressure, power in zip(flow_lps, pressure_kpa, power_kw):
    hydraulic_kw = flow * pressure / 1000.0
    efficiency = hydraulic_kw / power
    efficiencies.append(efficiency)

# Find the best operating point safely.
best_index = max(range(row_count), key=efficiencies.__getitem__)
best_efficiency = efficiencies[best_index]
best_flow = flow_lps[best_index]

# Write a readable formatted report.
with open(report_path, "w", encoding="utf-8") as file:
    file.write("Pump Performance Summary\n")
    file.write("Project: Training example\n")

    file.write("\nFlow     Pressure   Power    Efficiency\n")
    file.write("L/s      kPa        kW       percent\n")

    for flow, pressure, power, efficiency in zip(flow_lps, pressure_kpa, power_kw, efficiencies):
        file.write(f"{flow:6.1f} {pressure:9.1f} {power:7.2f} {efficiency * 100:10.1f}\n")

    file.write("\nKey result\n")
    file.write(f"Best efficiency: {best_efficiency * 100:.1f}%\n")
    file.write(f"Best flow rate: {best_flow:.1f} L/s\n")

# Read back a short preview.
with open(report_path, "r", encoding="utf-8") as file:
    preview_lines = file.readlines()[:8]

# Print only a compact preview.
print("Report file created:", report_path)
print("Preview of formatted output:")
print("".join(preview_lines))

```

>* Turn results into readable, trusted reports

>* Use consistent labels, spacing, and units

>* Make reports readable and predictable

```
#@title Python Code - Report Formatting

# Format engineering reports for readable output.
# Include labels, units, and precision.
# Write a small report file safely.

from pathlib import Path
from datetime import date

# Store compact beam calculation results.
results = [
    ("B1", 2.346, 5.000, "Pass"),
    ("B2", 5.284, 5.000, "Review"),
]

# Choose a safe local report path.
report_path = Path("beam_deflection_report.txt")

# Build clear report lines first.
lines = [
    "Beam Deflection Check Report",
    f"Date: {date.today().isoformat()}",
    "Units: deflection values are millimeters",
]

# Add a predictable table header.
lines.extend([
    "",
    "Beam     Max Defl.    Limit      Status",
    "ID       (mm)         (mm)       Decision",
])

# Format each result consistently.
for beam_id, deflection, limit, status in results:
    line = f"{beam_id:<8}{deflection:>8.2f}{limit:>10.2f}   {status}"
    lines.append(line)

# Add a concise engineering summary.
review_count = sum(status != "Pass" for *_, status in results)
lines.append("")
lines.append(f"Summary: {review_count} beam requires review.")

# Write the report using safe text settings.
try:
    report_path.write_text("\n".join(lines), encoding="utf-8")
except OSError as error:
    print(f"Could not write report: {error}")
else:
    print(f"Report written to {report_path}")
    print(report_path.read_text(encoding="utf-8"))

```

## **3. Error Handling**

### **3.1. Exception Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1780718599" width="250">



>* Exceptions signal unexpected runtime problems.
>* Real-world files and data commonly cause them.

>* Exceptions report problems in an organized way
>* Handlers guide users toward appropriate fixes

>* Match each error to the right response
>* Make failures visible before results become misleading



```
#@title Python Code - Exception Basics

# Exceptions help programs handle runtime problems.
# Engineering files often contain unexpected values.
# Defensive checks keep calculations trustworthy.

# Create a tiny CSV-style data file.
file_path = "beam_readings.csv"

# Store one valid row and one invalid row.
rows = ["span_m,load_n\n", "2.0,500\n", "0,300\n", "bad,700\n"]

# Write sample data for this demonstration.
with open(file_path, "w", encoding="utf-8") as file:
    file.writelines(rows)

# Prepare containers for results and errors.
stresses = []
errors = []

# Read the file using a protected block.
try:
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

except FileNotFoundError as problem:
    errors.append(f"Missing file: {problem.filename}")

# Process rows only when data exists.
else:
    for row_number, line in enumerate(lines[1:], start=2):
        parts = line.strip().split(",")

        # Check the expected column count.
        if len(parts) != 2:
            errors.append(f"Row {row_number}: wrong column count")
            continue

        # Convert text safely into numbers.
        try:
            span = float(parts[0])
            load = float(parts[1])

        except ValueError:
            errors.append(f"Row {row_number}: nonnumeric value found")
            continue

        # Prevent division by zero before calculating.
        if span == 0:
            errors.append(f"Row {row_number}: span cannot be zero")
            continue

        # Calculate a simple load per span value.
        stresses.append(load / span)

# Write a short report with clear formatting.
report_path = "error_handling_report.txt"
with open(report_path, "w", encoding="utf-8") as report:
    report.write("Error Handling Report\n")
    report.write(f"Valid calculations: {len(stresses)}\n")
    report.write(f"Issues found: {len(errors)}\n")

# Show concise results for the lecture.
print("Exception basics demonstration complete.")
print(f"Valid load/span values: {stresses}")
print(f"Handled issues: {errors}")
print(f"Report saved as: {report_path}")

```

### **3.2. Exception Handling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1780718602" width="250">



>* Plan for file and data problems
>* Respond clearly without crashing abruptly

>* Separate normal steps from recovery actions
>* Handle each error type deliberately

>* Match error responses to specific problems
>* Report issues clearly without hiding failures



```
#@title Python Code - Exception Handling

# Demonstrate exception handling for engineering files.
# Create small files inside Colab.
# Report errors without stopping abruptly.

# No extra packages are required.
from pathlib import Path
import csv
import statistics

# Define tiny file locations.
good_path = Path("sensor_readings.csv")
missing_path = Path("missing_readings.csv")
report_path = Path("temperature_report.txt")

# Write sample data with one bad row.
good_path.write_text(
    "time_s,temp_C\n0,21.5\n1,21.7\n2,error\n3,22.1\n",
    encoding="utf-8",
)

# Create a reusable CSV reading function.
def read_temperatures(path):
    temperatures = []
    skipped_rows = []

    # Handle file opening and parsing errors.
    try:
        with path.open("r", encoding="utf-8", newline="") as file:
            reader = csv.DictReader(file)

            # Check the expected column exists.
            if "temp_C" not in reader.fieldnames:
                raise ValueError("Missing required temp_C column")

            # Convert each row carefully.
            for row_number, row in enumerate(reader, start=2):
                try:
                    temperatures.append(float(row["temp_C"]))

                except ValueError:
                    skipped_rows.append(row_number)

    except FileNotFoundError:
        print(f"File not found: {path}")
        return [], []

    except PermissionError:
        print(f"Permission denied: {path}")
        return [], []

    except ValueError as problem:
        print(f"File format problem: {problem}")
        return [], []

    return temperatures, skipped_rows

# First show recovery from a missing file.
missing_values, missing_skips = read_temperatures(missing_path)
print(f"Missing-file read returned {len(missing_values)} values.")

# Now read the real file with bad data.
values, skipped = read_temperatures(good_path)

# Defensively verify useful data exists.
if not values:
    print("No valid temperatures were available.")

else:
    average_temp = statistics.mean(values)
    maximum_temp = max(values)

    # Save a concise engineering report safely.
    try:
        report_path.write_text(
            f"Valid readings: {len(values)}\n"
            f"Skipped rows: {skipped}\n"
            f"Average temperature: {average_temp:.2f} C\n"
            f"Maximum temperature: {maximum_temp:.2f} C\n",
            encoding="utf-8",
        )

    except OSError as problem:
        print(f"Could not save report: {problem}")

    else:
        print(f"Valid readings processed: {len(values)}")
        print(f"Rows skipped during parsing: {skipped}")
        print(f"Report saved as: {report_path}")

```

### **3.3. Defensive Error Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1780718605" width="250">



>* Check inputs before risky file operations
>* Verify data structure for safe processing

>* Silent data mistakes can mislead results
>* Defensive checks give clearer error messages

>* Check essential assumptions without cluttering code
>* Handle file changes while protecting results



```
#@title Python Code - Defensive Error Checks

# Defensive checks prevent confusing file errors.
# Small data files can still contain problems.
# Clear messages help engineers correct inputs.

# Create example CSV text with one bad row.
example_lines = [
    "time,temp,pressure,flow",
    "0.0,22.5,101.3,4.8",

    "1.0,23.1,101.1,4.9",
    "2.0,,100.9,5.0",
    "3.0,24.0,100.7,-1.0",
]

# Choose a safe local file path.
file_path = "sample_engineering_data.csv"

# Write the tiny example file.
with open(file_path, "w", encoding="utf-8") as file_object:
    file_object.write("\n".join(example_lines))

# Check the file path before opening.
if not file_path or not file_path.endswith(".csv"):
    raise ValueError("Expected a nonempty .csv file path.")

# Prepare containers for accepted rows.
valid_rows = []
rejected_rows = []
expected_columns = 4

# Read the file using safe handling.
try:
    with open(file_path, "r", encoding="utf-8") as file_object:
        lines = file_object.readlines()
except FileNotFoundError:
    lines = []
    rejected_rows.append("File was not found.")

# Check that data was actually loaded.
if len(lines) < 2:
    rejected_rows.append("File has no data rows.")
else:
    header = lines[0].strip().split(",")

# Validate the header before processing.
if len(lines) >= 2 and len(header) != expected_columns:
    rejected_rows.append("Header column count is wrong.")

# Process rows only when structure is plausible.
for row_number, line in enumerate(lines[1:], start=2):
    fields = line.strip().split(",")

    # Reject blank or incomplete rows early.
    if len(fields) != expected_columns or "" in fields:
        rejected_rows.append(f"Row {row_number}: missing field.")
        continue

    # Convert values only after simple checks.
    try:
        time, temp, pressure, flow = map(float, fields)
    except ValueError:
        rejected_rows.append(f"Row {row_number}: nonnumeric value.")
        continue

    # Reject physically unreasonable flow values.
    if flow < 0:
        rejected_rows.append(f"Row {row_number}: negative flow.")
        continue

    # Store validated engineering measurements.
    valid_rows.append((time, temp, pressure, flow))

# Compute a result only with valid data.
if valid_rows:
    average_temp = sum(row[1] for row in valid_rows) / len(valid_rows)
else:
    average_temp = None

# Print a compact report for learners.
print("Defensive file check report")
print(f"Valid rows accepted: {len(valid_rows)}")
print(f"Rows rejected: {len(rejected_rows)}")
print(f"Average valid temperature: {average_temp:.2f} C")
print("First issue:" , rejected_rows[0] if rejected_rows else "none")

```
