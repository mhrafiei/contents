## **1. Variables and Types**

### **1.1. Variable Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1780366199" width="250">



>* Variables name stored engineering values clearly
>* Clear names make calculations easier to understand

>* Assignment stores values under clear variable names
>* Variables make later calculations easier to understand

>* Variables store changing values during calculations
>* Clear assignments make programs readable and meaningful



```python
#@title Python Code - Variable Basics

# Variables store named engineering values.
# Assignment connects names with measurements.
# Clear names make formulas readable.

# Store measured pipe dimensions.
pipe_length_m = 12.0
pipe_diameter_m = 0.25

# Store material and flow quantities.
water_density_kg_m3 = 997.0
flow_speed_m_s = 1.8

# Compute cross sectional area.
pipe_area_m2 = 3.14159 * (pipe_diameter_m / 2) ** 2

# Compute volume flow rate.
flow_rate_m3_s = pipe_area_m2 * flow_speed_m_s

# Compute mass flow rate.
mass_flow_rate_kg_s = flow_rate_m3_s * water_density_kg_m3

# Update length for revised design.
pipe_length_m = 15.0

# Print labeled calculation results.
print(f"Pipe length: {pipe_length_m:.1f} m")
print(f"Pipe diameter: {pipe_diameter_m:.2f} m")
print(f"Pipe area: {pipe_area_m2:.4f} m^2")
print(f"Volume flow rate: {flow_rate_m3_s:.4f} m^3/s")
print(f"Mass flow rate: {mass_flow_rate_kg_s:.2f} kg/s")

```

### **1.2. Variable Types**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1780366201" width="250">



>* Types describe values and valid operations
>* Engineering data includes counts, measurements, labels, decisions

>* Integers count; floats measure decimal quantities
>* Floats are approximate, so use judgment

>* Text adds context to engineering results
>* Booleans capture decisions and design checks



```python
#@title Python Code - Variable Types

# Variables store different engineering information.
# Types describe values and operations.
# Clear types make calculations readable.

# Store an exact whole count.
bolt_count = int(8)

# Store measured decimal engineering quantities.
beam_length_m = float(6.5)
load_per_m_kn = float(12.0)

# Store descriptive text values.
member_name = str("Roof beam B2")
force_unit = str("kN")

# Store a true-or-false design condition.
allowable_load_kn = float(90.0)
total_load_kn = beam_length_m * load_per_m_kn

# Compare measurement against a limit.
design_passes = total_load_kn <= allowable_load_kn

# Print each value with its type.
print(f"Member: {member_name} ({type(member_name).__name__})")
print(f"Bolt count: {bolt_count} ({type(bolt_count).__name__})")
print(f"Beam length: {beam_length_m} m ({type(beam_length_m).__name__})")
print(f"Total load: {total_load_kn:.1f} {force_unit} ({type(total_load_kn).__name__})")
print(f"Design passes: {design_passes} ({type(design_passes).__name__})")

```

## **2. Engineering Arithmetic**

### **2.1. Operator Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780366197" width="250">



>* Operators turn engineering quantities into results
>* Choose operations to match the situation

>* Follow operation order in mixed calculations
>* Use grouping to show intended formulas

>* Interpret operators based on engineering context
>* Connect calculations to real system meaning



```python
#@title Python Code - Operator Basics

# Arithmetic operators support engineering formulas.
# Parentheses make calculation intent explicit.
# Printed results connect numbers to units.

# Define basic engineering quantities.
beam_length_m = 6.0
uniform_load_n_per_m = 850.0
support_count = 2

# Use multiplication for total applied load.
total_load_n = beam_length_m * uniform_load_n_per_m
load_per_support_n = total_load_n / support_count

# Use powers for area-based formulas.
radius_m = 0.012
cross_section_area_m2 = 3.14159 * radius_m ** 2

# Use floor division for full pieces.
stock_length_m = 5.0
piece_length_m = 1.2
full_pieces = stock_length_m // piece_length_m

# Use modulo for remaining material.
leftover_m = stock_length_m % piece_length_m

# Show why grouping changes meaning.
wrong_average_n = total_load_n + 200 / support_count
correct_average_n = (total_load_n + 200) / support_count

# Print concise engineering results.
print(f"Total load: {total_load_n:.1f} N")
print(f"Load per support: {load_per_support_n:.1f} N")
print(f"Rod area: {cross_section_area_m2:.6f} m^2")
print(f"Full pieces cut: {full_pieces:.0f}")
print(f"Leftover stock: {leftover_m:.2f} m")
print(f"Without grouping: {wrong_average_n:.1f} N")
print(f"With grouping: {correct_average_n:.1f} N")

```

### **2.2. Built In Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780366203" width="250">



>* Use built-ins for common engineering calculations
>* Round and measure results clearly

>* Make calculation goals easy to understand
>* Reduce repeated logic and improve reliability

>* Turn raw data into engineering insight
>* Make formulas clearer and easier to verify



```python
#@title Python Code - Built In Functions

# Built-in functions simplify common engineering calculations.
# This example summarizes simple test measurements.
# Results are rounded for readable reporting.

# Store measured and target temperatures.
target_temperature_c = 80.0
measured_temperature_c = 76.8

# Calculate signed error using subtraction.
temperature_error_c = measured_temperature_c - target_temperature_c

# Use abs to find error magnitude.
error_magnitude_c = abs(temperature_error_c)

# Store several beam load test results.
load_tests_kn = [12.4, 15.9, 14.7, 16.2]

# Use max and min for extremes.
maximum_load_kn = max(load_tests_kn)
minimum_load_kn = min(load_tests_kn)

# Use sum and len for average.
average_load_kn = sum(load_tests_kn) / len(load_tests_kn)

# Round values for engineering output.
rounded_error_c = round(error_magnitude_c, 2)
rounded_average_kn = round(average_load_kn, 2)

# Print a compact engineering summary.
print("Temperature error magnitude:", rounded_error_c, "deg C")
print("Maximum tested load:", maximum_load_kn, "kN")
print("Minimum tested load:", minimum_load_kn, "kN")
print("Average tested load:", rounded_average_kn, "kN")

```

### **2.3. Built In Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780366196" width="250">



>* Use built-ins for common engineering calculations
>* Make formulas clearer and less error-prone

>* Use abs, max, and min for comparisons
>* Round results for clear engineering communication

>* Respect units, precision, and measurement limits
>* Use functions for clear, reliable calculations



```python
#@title Python Code - Built In Functions

# Demonstrate built in engineering calculation tools.
# Use measurements with clear engineering units.
# Keep output concise and interpretable.

# Store recorded beam loads in kilonewtons.
loads_kN = [18.4, 21.7, 19.9, 23.2]

# Store target and measured diameters in millimeters.
target_mm = 50.0
measured_mm = 49.63

# Use max and min for recorded bounds.
peak_load_kN = max(loads_kN)
lowest_load_kN = min(loads_kN)

# Use abs for unsigned manufacturing error.
diameter_error_mm = abs(measured_mm - target_mm)

# Use round for reporting sensible precision.
reported_peak = round(peak_load_kN, 1)
reported_error = round(diameter_error_mm, 2)

# Print concise engineering results with units.
print(f"Peak load: {reported_peak} kN")
print(f"Lowest load: {lowest_load_kN:.1f} kN")
print(f"Diameter error: {reported_error} mm")
print(f"Number of tests: {len(loads_kN)}")

```

## **3. Readable Output**

### **3.1. Print Output**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780366208" width="250">



>* Print messages give calculation results clear meaning
>* Descriptive output improves testing and trust

>* Label printed values for human understanding
>* Use messages to compare and debug results

>* Match print detail to audience and purpose
>* Use output to explain technical reasoning



```python
#@title Python Code - Print Output

# Print statements communicate engineering calculation results.
# Labels make numeric outputs easier to understand.
# Units help readers interpret computed values.

# Store clear engineering quantities as variables.
beam_length_m = 3.0
load_newtons = 450.0
stiffness_newtons_per_m = 18000.0

# Compute a simple deflection estimate.
deflection_m = load_newtons / stiffness_newtons_per_m
deflection_mm = deflection_m * 1000.0
safety_limit_mm = 30.0

# Print raw output for comparison.
print("Raw value:", deflection_mm)

# Print readable output with labels and units.
print(f"Beam length: {beam_length_m:.1f} m")
print(f"Applied load: {load_newtons:.0f} N")
print(f"Estimated deflection: {deflection_mm:.2f} mm")

# Print a clear engineering conclusion.
within_limit = deflection_mm <= safety_limit_mm
print(f"Deflection limit: {safety_limit_mm:.1f} mm")
print(f"Within allowable limit: {within_limit}")

```

### **3.2. Formatted Strings**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780366210" width="250">



>* Combine text and values clearly
>* Add context, labels, and sensible precision

>* Control precision for readable engineering results
>* Keep repeated outputs consistent and scannable

>* Connect results to real-world decisions
>* Make calculations clear, professional, and reliable



```python
#@title Python Code - Formatted Strings

# Formatted strings make engineering results readable.
# Units and labels prevent calculation confusion.
# Rounded values communicate appropriate practical precision.

# Store clear engineering input values.
beam_length_m = 3.0
load_newton = 750.0
stiffness_newton_per_m = 28000.0

# Compute a simple deflection estimate.
deflection_m = load_newton / stiffness_newton_per_m
deflection_mm = deflection_m * 1000.0
percent_length = deflection_m / beam_length_m * 100.0

# Print raw output for comparison.
print("Raw deflection value:", deflection_mm)

# Format values with labels and units.
print(f"Beam length: {beam_length_m:.2f} m")
print(f"Applied load: {load_newton:.0f} N")
print(f"Estimated deflection: {deflection_mm:.2f} mm")
print(f"Deflection ratio: {percent_length:.3f}% of beam length")

# Format a short engineering conclusion.
summary = f"Under {load_newton:.0f} N, deflection is {deflection_mm:.2f} mm."
print(summary)

```

### **3.3. Unit Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780366212" width="250">



>* Units give numbers clear physical meaning
>* Labels prevent confusion in shared results

>* Combine labels, values, and units clearly
>* Label multiple results to avoid confusion

>* Use consistent units and clear conversion labels
>* Readable labels improve safety, trust, and professionalism



```python
#@title Python Code - Unit Labels

# Unit labels make engineering results readable.
# Clear output reduces interpretation mistakes.
# This example formats values with units.

# Store engineering quantities with descriptive variable names.
beam_deflection_mm = 2.347
flow_rate_lpm = 18.6
pressure_rise_kpa = 245.2

# Store efficiency as a fraction for calculation.
efficiency_fraction = 0.783
shaft_power_kw = 1.42
heat_rate_watts = 865.4

# Print a plain number to show ambiguity.
print("Without unit labels: 2.347")

# Print labeled values using formatted string literals.
print(f"Beam deflection: {beam_deflection_mm:.2f} mm")
print(f"Flow rate: {flow_rate_lpm:.1f} L/min")
print(f"Pressure rise: {pressure_rise_kpa:.1f} kPa")

# Convert efficiency fraction into percent.
efficiency_percent = efficiency_fraction * 100

# Print remaining results with clear unit labels.
print(f"Shaft power: {shaft_power_kw:.2f} kW")
print(f"Heat transfer rate: {heat_rate_watts:.0f} W")
print(f"Pump efficiency: {efficiency_percent:.1f} %")

```
