### **1.1. Threshold Decisions**

```
#@title Python Code - Threshold Decisions

# Model engineering thresholds with Python conditionals.
# Compare measured values against operating limits.
# Print concise classifications for review.

# Store example temperatures in degrees Celsius.
temperatures_c = [62.0, 78.5, 85.0, 91.2]

# Define inclusive warning and shutdown limits.
warning_limit_c = 80.0
shutdown_limit_c = 90.0

# Validate threshold ordering before comparisons.
valid_limits = warning_limit_c < shutdown_limit_c

# Stop early if limits are unsafe.
if not valid_limits:
    raise ValueError("Warning limit must stay below shutdown limit")

# Display the decision rule once.
print("Motor temperature threshold decisions:")

# Check each measurement against thresholds.
for temperature_c in temperatures_c:
    if temperature_c >= shutdown_limit_c:
        status = "SHUTDOWN"
    elif temperature_c >= warning_limit_c:
        status = "WARNING"
    else:
        status = "NORMAL"

    # Print one concise engineering decision.
    print(f"{temperature_c:5.1f} C -> {status}")

```

```
#@title Python Code - Threshold Logic

# Threshold logic turns limits into decisions.
# Engineering boundaries need careful comparison choices.
# This example classifies battery temperatures.

temperatures_c = [18.0, 34.5, 40.0, 45.0, 52.0]
normal_limit_c = 40.0
warning_limit_c = 50.0

# Validate limits before classifying measurements.
if normal_limit_c >= warning_limit_c:
    raise ValueError("Normal limit must stay below warning limit.")

# Classify each reading using ordered thresholds.
for temperature_c in temperatures_c:
    if temperature_c < normal_limit_c:
        status = "normal operation"
    elif temperature_c <= warning_limit_c:
        status = "warning zone"

    else:
        status = "stop charging"
    message = f"{temperature_c:4.1f} C -> {status}"
    print(message)

# Show the exact boundary interpretation.
print("At 40.0 C, the pack enters the warning zone.")

```

### **1.3. Limit Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1780713928" width="250">



>* Compare values against safe operating limits
>* Choose warnings, rejection, or continued calculation

>* Match checks to real engineering meaning
>* Separate safe, warning, and unsafe cases

>* Question suspicious engineering data before trusting it
>* Use conditionals for safer, credible results



```
#@title Python Code - Limit Checks

# Limit checks protect engineering calculations.
# Conditional statements encode safe operating rules.
# This example classifies pump flow readings.

# Define acceptable and warning flow limits.
low_limit = 20.0
normal_limit = 80.0
high_limit = 100.0

# Store small measured flow examples.
flow_values = [12.0, 45.0, 88.0, 115.0]
status_counts = {"unsafe": 0, "normal": 0, "warning": 0}

# Check each flow against engineering limits.
for flow_rate in flow_values:
    if flow_rate < low_limit or flow_rate > high_limit:
        status = "unsafe"

    elif flow_rate <= normal_limit:
        status = "normal"
    else:
        status = "warning"

    status_counts[status] = status_counts[status] + 1
    print(f"Flow {flow_rate:5.1f} L/min -> {status}")

# Summarize the limit check decisions.
print("Summary counts:", status_counts)

```

## **2. Engineering Loops**

### **2.1. Force Iteration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780713930" width="250">



>* Repeat force calculations across changing input values
>* Use loops for consistent range analysis

>* Loops automate repeated force calculations
>* Results reveal patterns for engineering decisions

>* Test forces across realistic operating ranges
>* Spot limits for safer design decisions



```
#@title Python Code - Force Iteration

# Iterate forces across controlled engineering values.
# Loops reveal trends across changing loads.
# Validation prevents unsafe calculation inputs.

# Define a safe spring constant.
spring_constant = 250.0

# Define displacement values in meters.
displacements = [0.00, 0.01, 0.02, 0.03, 0.04]

# Define the operating force limit.
safe_force_limit = 8.0

# Validate the spring constant before looping.
if spring_constant <= 0:
    raise ValueError("Spring constant must be positive.")

# Validate the displacement list before looping.
if len(displacements) == 0:
    raise ValueError("Displacement list cannot be empty.")

# Print a compact table heading.
print("Displacement(m)  Force(N)  Status")

# Repeat the force calculation for each displacement.
for displacement in displacements:

    # Reject negative compression distances.
    if displacement < 0:
        status = "invalid displacement"

    # Calculate force using Hooke's law.
    else:
        force = spring_constant * displacement

        # Compare force against the operating limit.
        if force <= safe_force_limit:
            status = "within limit"

        # Mark forces beyond the safe limit.
        else:
            status = "above limit"

        # Print one result line per case.
        print(f"{displacement:14.2f}  {force:8.2f}  {status}")

# Summarize the engineering lesson.
print("Looping shows how force changes across a range.")

```

### **2.2. While Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780713935" width="250">



>* Repeat while an engineering condition remains true
>* Stop when the result reaches its limit

>* Update loop values toward completion
>* Use clear limits and stopping rules

>* Validate inputs before calculations continue
>* Repeat until system behavior meets limits



```
#@title Python Code - While Loops

# Model tank draining with while loops.
# Stop when concentration reaches safe limit.
# Track each controlled calculation step.

# Set starting engineering values.
concentration_mg_L = 120.0
safe_limit_mg_L = 25.0
removal_fraction = 0.30

# Add a safety limit for iterations.
step_count = 0
max_steps = 20
history = []

# Repeat while concentration remains unsafe.
while concentration_mg_L > safe_limit_mg_L and step_count < max_steps:
    history.append(round(concentration_mg_L, 2))
    concentration_mg_L *= (1 - removal_fraction)
    step_count += 1

# Store the final calculated concentration.
final_value = round(concentration_mg_L, 2)
history.append(final_value)
status_safe = concentration_mg_L <= safe_limit_mg_L

# Report concise engineering loop results.
print("While loop tank cleanup example")
print(f"Starting concentration: {history[0]} mg/L")
print(f"Safe discharge limit: {safe_limit_mg_L} mg/L")

# Show the controlled repeated calculations.
print(f"Treatment steps completed: {step_count}")
print(f"Final concentration: {final_value} mg/L")
print(f"Limit reached safely: {status_safe}")

```

### **2.3. Looping Engineering Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780713936" width="250">



>* Repeat calculations across changing engineering conditions
>* Reveal trends, limits, and safety margins

>* Choose ranges and steps carefully
>* Compare options with fewer repeat errors

>* Use loops to explore design behavior
>* Reveal patterns, limits, and sensitivities



```
#@title Python Code - Looping Engineering Values

# Explore loads using a simple loop.
# Repeat one beam stress calculation safely.
# Compare each result with an allowable limit.

# Define a small engineering value sweep.
loads_kn = [5, 10, 15, 20, 25, 30]

# Store beam and material constants.
area_mm2 = 250.0
allowable_mpa = 80.0

# Validate constants before repeated calculations.
if area_mm2 <= 0.0:
    raise ValueError("Area must be positive.")

# Print a compact heading for results.
print("Load sweep for axial stress")

# Loop through each engineering load value.
for load_kn in loads_kn:
    if load_kn < 0.0:
        continue

    # Convert kilonewtons to newtons.
    force_n = load_kn * 1000.0

    # Stress in MPa equals newtons per square millimeter.
    stress_mpa = force_n / area_mm2

    # Classify each operating condition.
    if stress_mpa <= allowable_mpa:
        status = "OK"
    else:
        status = "LIMIT EXCEEDED"

    # Print one clear line per case.
    print(f"{load_kn:2.0f} kN -> {stress_mpa:6.1f} MPa, {status}")

```

## **3. Validation Logic**

### **3.1. Safe Input Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780713943" width="250">



>* Check inputs before running engineering calculations
>* Reject missing, wrong, or impossible values

>* Check inputs are usable before calculating
>* Catch errors early for clearer feedback

>* Reveal calculation limits before trusting results
>* Validate inputs while using engineering judgment



```
#@title Python Code - Safe Input Checks

# Safe checks protect engineering calculations.
# Invalid values should stop early.
# Clear messages guide safer corrections.

# Store sample entries from simulated sources.
raw_entries = ["0.05", "-0.10", "abc", ""]
accepted_values = []
rejected_messages = []

# Validate each entry before calculation.
for position, entry in enumerate(raw_entries, start=1):
    cleaned_entry = entry.strip()
    if cleaned_entry == "":
        rejected_messages.append(f"entry {position}: missing value")
        continue

    # Convert text only after presence checks.
    try:
        diameter_m = float(cleaned_entry)
    except ValueError:
        rejected_messages.append(f"entry {position}: not numeric")
        continue

    # Reject physically impossible diameters.
    if diameter_m <= 0:
        rejected_messages.append(f"entry {position}: not positive")
        continue
    accepted_values.append(diameter_m)

# Use validated values in later calculations.
for diameter_m in accepted_values:
    area_m2 = 3.14159 * (diameter_m / 2) ** 2
    print(f"accepted diameter {diameter_m:.3f} m gives area {area_m2:.5f} m^2")

# Report rejected inputs without crashing.
print(f"accepted count: {len(accepted_values)}")
print(f"rejected count: {len(rejected_messages)}")
print("reasons: " + "; ".join(rejected_messages))

```

### **3.2. Input Range Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780713941" width="250">



>* Check inputs against realistic engineering limits
>* Prevent unsafe results from invalid assumptions

>* Limits come from physics, equipment, and standards
>* Clear boundaries show when results are trustworthy

>* Show clear messages for out-of-range inputs
>* Prevent unsafe calculations and support engineering judgment



```
#@title Python Code - Input Range Checks

# Range checks protect engineering calculations.
# Limits should match physical assumptions.
# Clear messages help users correct values.

# Store sample sensor readings in percent.
state_of_charge_values = [-5, 0, 42, 100, 115]

# Define the valid engineering range.
lower_limit, upper_limit = 0, 100

# Prepare counters for validation results.
valid_count, invalid_count = 0, 0

# Check each value before using it.
for state_of_charge in state_of_charge_values:
    # Test whether the reading is inside limits.
    is_valid = lower_limit <= state_of_charge <= upper_limit

    # Report clear validation feedback.
    if is_valid:
        valid_count += 1

        # Use only validated inputs in calculations.
        energy_factor = state_of_charge / upper_limit
        print(f"{state_of_charge}% is valid; energy factor = {energy_factor:.2f}")

    else:
        invalid_count += 1

        # Explain why the input is rejected.
        print(f"{state_of_charge}% is outside the safe 0 to 100 range.")

# Summarize the validation process briefly.
print(f"Valid readings: {valid_count}; rejected readings: {invalid_count}")

```

### **3.3. Handling Unsafe Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780713939" width="250">



>* Reject inputs that break engineering assumptions
>* Show errors before unsafe results appear

>* Stop calculations before unsafe results appear
>* Explain input problems clearly for correction

>* Choose retry or defaults carefully
>* Log unsafe inputs to spot patterns



```
#@title Python Code - Handling Unsafe Inputs

# Validate inputs before engineering calculations.
# Unsafe values should stop results.
# Clear messages guide safer corrections.

# Store sample pipe diameters in meters.
diameters_m = [0.25, -0.10, 0.00, 0.15]

# Store sample flow rates in cubic meters per second.
flow_rates_m3s = [0.020, 0.015, 0.010, -0.005]

# Define safe engineering limits.
minimum_diameter_m = 0.05
maximum_flow_m3s = 0.050

# Prepare containers for accepted calculations.
accepted_speeds = []
rejected_cases = []

# Check each paired engineering input.
for case_number, values in enumerate(zip(diameters_m, flow_rates_m3s), start=1):
    diameter_m, flow_m3s = values

    # Reject impossible or unsafe values.
    if diameter_m < minimum_diameter_m:
        rejected_cases.append((case_number, "diameter outside safe range"))
        continue

    # Reject negative or excessive flow rates.
    if flow_m3s <= 0 or flow_m3s > maximum_flow_m3s:
        rejected_cases.append((case_number, "flow outside safe range"))
        continue

    # Calculate area after validation passes.
    pipe_area_m2 = 3.14159 * (diameter_m / 2) ** 2
    speed_mps = flow_m3s / pipe_area_m2
    accepted_speeds.append((case_number, round(speed_mps, 2)))

# Report only trusted calculation results.
print("Validated pipe-flow speed results:")
print("Accepted cases:", accepted_speeds)
print("Rejected cases:", rejected_cases)

# Summarize why validation matters.
print("Unsafe inputs were rejected before producing misleading speeds.")

```
