## **1. Ordered Engineering Data**

### **1.1. List Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780714901" width="250">



>* Lists store engineering values in meaningful order
>* Sequence preserves time, location, and patterns

>* Lists adapt as data changes
>* They store varied, updateable engineering information

>* Define what each list position means
>* Keep order tied to real systems



```
#@title Python Code - List Basics

# Lists store ordered engineering data.
# Positions preserve measurement sequence meaning.
# Built-ins summarize values quickly.

# Create sequential pressure readings.
pressure_kpa = [101.2, 101.8, 102.5, 103.1, 102.7]

# Create ordered sensor labels.
sensor_names = ["inlet", "midpipe", "outlet"]

# Access first and last readings.
first_reading = pressure_kpa[0]
last_reading = pressure_kpa[-1]

# Slice the middle three readings.
middle_readings = pressure_kpa[1:4]

# Append a new measured reading.
pressure_kpa.append(103.4)

# Compute simple summary values.
average_pressure = sum(pressure_kpa) / len(pressure_kpa)
pressure_range = max(pressure_kpa) - min(pressure_kpa)

# Print compact teaching results.
print("Pressure readings:", pressure_kpa)
print("Sensor order:", sensor_names)
print("First and last:", first_reading, last_reading)
print("Middle readings:", middle_readings)
print("Average pressure:", round(average_pressure, 2), "kPa")
print("Pressure range:", round(pressure_range, 2), "kPa")

```

### **1.2. Tuple Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780714905" width="250">



>* Tuples group related engineering values in order
>* Positions give each record element meaning

>* Tuples keep records fixed and protected
>* Use lists for data that changes

>* Use tuples for clear engineering records
>* Define positions and fixed sequence meaning



```
#@title Python Code - Tuple Data

# Tuples store ordered engineering records.
# Positions in tuples carry meaning.
# Tuple contents cannot change in place.

# Create fixed coordinate and sensor records.
point_xyz = (12.5, 3.8, 1.2)
sensor_record = ("T-17", "2026-06-06", 84.6)
material_props = (7850, 250, 45)

# Access values by their stable positions.
x_value = point_xyz[0]
y_value = point_xyz[1]
elevation_value = point_xyz[2]

# Unpack one tuple into named variables.
sensor_id, date_text, temperature_c = sensor_record
record_count = len(sensor_record)

# Use built-ins for compact summary results.
coordinate_sum = sum(point_xyz)
average_coordinate = coordinate_sum / len(point_xyz)
strongest_value = max(material_props)

# Display concise engineering tuple results.
print("Coordinate tuple:", point_xyz)
print("X, Y, elevation:", x_value, y_value, elevation_value)
print("Sensor reading:", sensor_id, date_text, temperature_c)
print("Sensor fields:", record_count)
print("Average coordinate value:", round(average_coordinate, 2))
print("Largest material property:", strongest_value)

```

### **1.3. Named Data Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780714907" width="250">



>* Names clarify positions in ordered data
>* Improves accuracy when sharing or analyzing

>* Names reduce ordering mistakes in engineering records
>* Named values clarify raw measurements across fields

>* Names make ordered data easier to understand
>* Clear records support reliable team workflows



```
#@title Python Code - Named Data Sequences

# Named sequences label ordered engineering records.
# They preserve positions while improving readability.
# This example uses built-in namedtuple records.

# Import namedtuple from the built-in collections module.
from collections import namedtuple

# Define meaningful names for each ordered field.
SensorReading = namedtuple(
    "SensorReading", "temperature pressure flow_rate timestamp"
)

# Create a small ordered engineering data record.
reading = SensorReading(72.5, 101.3, 4.8, "09:00")

# Show that indexing still works with positions.
indexed_temperature = reading[0]

# Show that names make positions easier to understand.
named_temperature = reading.temperature

# Store several named records inside a list.
readings = [
    SensorReading(72.5, 101.3, 4.8, "09:00"),
    SensorReading(73.1, 101.1, 5.0, "09:05"),
    SensorReading(72.8, 101.4, 4.9, "09:10"),
]

# Validate that the list contains expected records.
assert len(readings) == 3 and len(readings[0]) == 4

# Compute a summary using clear field names.
average_flow = sum(item.flow_rate for item in readings) / len(readings)

# Print concise teaching results.
print("Record:", reading)
print("Temperature by index:", indexed_temperature)
print("Temperature by name:", named_temperature)
print("Average flow rate:", round(average_flow, 2))
print("Named fields:", SensorReading._fields)

```

## **2. Access Patterns**

### **2.1. Indexing Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780714892" width="250">



>* Indexes locate values in ordered sequences
>* Zero-based positions select specific engineering data

>* Index readings by their sequence position
>* Positions connect data to real processes

>* Use negative indexes for end positions
>* Stay within valid sequence index ranges



```
#@title Python Code - Indexing Basics

# Lists store ordered engineering measurements.
# Indexes retrieve one sequence value.
# Negative indexes count from endings.

# Store five chronological temperature readings.
temperature_c = [21.5, 22.0, 22.8, 23.1, 23.7]

# Store matching elapsed time labels.
time_minutes = (0, 5, 10, 15, 20)

# Check both sequences share length.
if len(temperature_c) != len(time_minutes):
    raise ValueError("Sequence lengths must match.")

# Access baseline using index zero.
baseline_temp = temperature_c[0]
baseline_time = time_minutes[0]

# Access third reading using index two.
third_temp = temperature_c[2]
third_time = time_minutes[2]

# Access final reading using negative indexing.
final_temp = temperature_c[-1]
final_time = time_minutes[-1]

# Compute change between first and last.
temperature_change = final_temp - baseline_temp

# Display selected indexed engineering values.
print("Temperature readings:", temperature_c)
print("Baseline:", baseline_temp, "C at", baseline_time, "min")
print("Third reading:", third_temp, "C at", third_time, "min")
print("Final reading:", final_temp, "C at", final_time, "min")
print("Final minus baseline:", round(temperature_change, 2), "C")

```

### **2.2. Slicing Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780714886" width="250">



>* Select connected ranges from ordered sequences
>* Analyze meaningful segments of engineering data

>* Start included, end excluded
>* Omit boundaries for common slice patterns

>* Use steps to sample large sequences
>* Create focused copies while preserving originals



```
#@title Python Code - Slicing Sequences

# Slicing selects meaningful parts from ordered data.
# Engineering readings often arrive in sequence order.
# Slice boundaries exclude the ending index.

# Store hourly thermal readings in Celsius.
temperatures = [21.0, 22.4, 24.1, 25.3, 25.7, 25.9, 25.6, 24.8]

# Create slices for common test phases.
warmup_readings = temperatures[:3]
stable_readings = temperatures[3:6]
shutdown_readings = temperatures[-2:]

# Select every second reading for quick review.
every_second_reading = temperatures[::2]

# Keep the original sequence unchanged.
original_count = len(temperatures)
slice_count = len(stable_readings)

# Compute summary values from sliced data.
stable_average = sum(stable_readings) / slice_count
quick_maximum = max(every_second_reading)

print("All readings:", temperatures)
print("Warm-up slice temperatures[:3]:", warmup_readings)
print("Stable slice temperatures[3:6]:", stable_readings)
print("Shutdown slice temperatures[-2:]:", shutdown_readings)
print("Stepped slice temperatures[::2]:", every_second_reading)
print("Original count remains:", original_count)
print("Stable average:", round(stable_average, 2))
print("Quick-review maximum:", quick_maximum)

```

### **2.3. Looping Through Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780714888" width="250">



>* Visit each sequence value in order
>* Process changing data sets consistently

>* Loop by value or by position
>* Use indexes when order affects meaning

>* Use loops to summarize sequence data
>* Scale analysis with indexing and slicing



```
#@title Python Code - Looping Through Sequences

# Looping visits sequence values in order.
# Engineering readings often need repeated processing.
# Indexes add position information when needed.

# Store hourly vibration readings in a list.
vibration_mm_s = [1.8, 2.1, 3.7, 2.9, 4.2, 2.4]

# Store matching hour labels in a tuple.
hours = (0, 1, 2, 3, 4, 5)

# Check paired sequence lengths before looping.
if len(vibration_mm_s) == len(hours):
    total_vibration = 0.0
    exceedance_count = 0

    # Loop with indexes and values together.
    for index, reading in enumerate(vibration_mm_s):
        total_vibration = total_vibration + reading
        hour_label = hours[index]

        # Count readings above the engineering threshold.
        if reading > 3.0:
            exceedance_count = exceedance_count + 1
            print(f"Hour {hour_label}: {reading} mm/s exceeds limit")

    # Compute a summary using accumulated values.
    average_vibration = total_vibration / len(vibration_mm_s)

    # Compare loop results with built-in helpers.
    maximum_vibration = max(vibration_mm_s)
    minimum_vibration = min(vibration_mm_s)

    # Print concise summary results.
    print(f"Average vibration: {average_vibration:.2f} mm/s")
    print(f"Minimum vibration: {minimum_vibration:.1f} mm/s")
    print(f"Maximum vibration: {maximum_vibration:.1f} mm/s")
    print(f"Readings above limit: {exceedance_count}")
else:
    print("Data error: readings and hours mismatch")

```

## **3. Sequence Summaries**

### **3.1. Aggregate Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780714903" width="250">



>* Summaries combine sequence data into key results
>* Aggregates support engineering decisions and comparisons

>* Loop through sequences, updating a running summary
>* Adapt the pattern for many summaries

>* Choose summaries that fit the question
>* Check assumptions and interpret results carefully



```
#@title Python Code - Aggregate Values

# Aggregate values summarize engineering sequence data.
# Loops build summaries one element each time.
# Builtins compute common summaries very directly.

# Store ordered temperature readings in Celsius.
temperatures_c = [21.4, 22.1, 24.8, 26.3, 25.7, 23.9]

# Start a running total before iteration.
running_total = 0.0

# Add each reading to the running total.
for reading in temperatures_c:
    running_total = running_total + reading

# Count readings above a safety threshold.
high_count = 0
threshold_c = 25.0

# Increase the count when condition is true.
for reading in temperatures_c:
    if reading > threshold_c:
        high_count = high_count + 1

# Compute builtin summaries for comparison.
builtin_total = sum(temperatures_c)
highest_reading = max(temperatures_c)
lowest_reading = min(temperatures_c)

# Validate sequence length before computing average.
reading_count = len(temperatures_c)
average_c = builtin_total / reading_count

# Print compact summary results for interpretation.
print("Temperature readings:", temperatures_c)
print("Loop total:", round(running_total, 2), "C")
print("Builtin total:", round(builtin_total, 2), "C")
print("Average:", round(average_c, 2), "C")
print("Lowest and highest:", lowest_reading, highest_reading)
print("Readings above threshold:", high_count)

```

### **3.2. Built In Summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780714899" width="250">



>* Summarize sequence data quickly and reliably
>* Find totals, extremes, and counts easily

>* Built-in summaries make program intent clear
>* They reduce errors and support decisions

>* Validate data before using summaries
>* Combine summaries with slicing for analysis



```
#@title Python Code - Built In Summaries

# Summaries quickly describe ordered engineering measurements.
# Built-in functions reduce manual loop mistakes.
# Lists and tuples both support summaries.

# Store hourly pump power readings.
pump_kw = [4.8, 5.1, 4.9, 5.4, 5.0, 5.2]

# Store fixed test load limits.
load_limits_kn = (12, 15, 18, 21)

# Compute summaries using built-in functions.
total_energy = sum(pump_kw)
reading_count = len(pump_kw)

# Continue summarizing important engineering values.
maximum_power = max(pump_kw)
minimum_power = min(pump_kw)

# Combine total and count safely.
average_power = total_energy / reading_count

# Summarize the tuple data too.
maximum_load = max(load_limits_kn)
minimum_load = min(load_limits_kn)

# Print compact sequence summary results.
print("Pump readings:", pump_kw)
print("Total energy estimate:", total_energy, "kWh")
print("Average power:", round(average_power, 2), "kW")

# Print important extreme values.
print("Minimum pump power:", minimum_power, "kW")
print("Maximum pump power:", maximum_power, "kW")
print("Load limit range:", minimum_load, "to", maximum_load, "kN")

```

### **3.3. Built In Summaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780714894" width="250">



>* Built-ins summarize sequences quickly and clearly
>* Totals, counts, minimums, and maximums guide decisions

>* Built-ins make summary code clearer
>* Quick checks reveal data issues

>* Match summaries to data type and meaning
>* Combine loops with summaries for insight



```
#@title Python Code - Built In Summaries

# Built in summaries describe sequence data.
# Engineering measurements often need quick summaries.
# Lists and tuples both support summaries.

# Store ordered readings in a list.
loads_kn = [12.5, 14.0, 13.2, 15.1, 11.8]

# Store fixed calibration values in a tuple.
strengths_mpa = (248, 255, 251, 260, 246)

# Validate data before numeric summaries.
if len(loads_kn) > 0 and len(strengths_mpa) > 0:
    total_load = sum(loads_kn)
    average_load = total_load / len(loads_kn)

    weakest_strength = min(strengths_mpa)
    strongest_strength = max(strengths_mpa)
else:
    raise ValueError("Sequences must contain measurements.")

# Print compact summary results.
print("Number of load readings:", len(loads_kn))
print("Total load in kN:", round(total_load, 2))
print("Average load in kN:", round(average_load, 2))

# Print tuple summary results.
print("Number of strength samples:", len(strengths_mpa))
print("Weakest strength in MPa:", weakest_strength)
print("Strongest strength in MPa:", strongest_strength)

# Check whether a safety threshold is reached.
threshold_mpa = 250
safe_samples = [value for value in strengths_mpa if value >= threshold_mpa]
print("Samples meeting threshold:", len(safe_samples))

```
