## **1. Dictionary Lookups**

### **1.1. Key Based Retrieval**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780715748" width="250">



>* Retrieve values using descriptive keys
>* Organize engineering data by meaning

>* Retrieve values by descriptive dictionary keys
>* Avoid fragile position-based engineering records

>* Clear keys describe stored engineering values
>* Explicit lookups organize data for decisions



```
#@title Python Code - Key Based Retrieval

# Dictionaries retrieve values using meaningful engineering keys.
# Key lookups avoid fragile position based access.
# This example stores one material record.

# Create a compact engineering property dictionary.
material_record = {
    "material": "6061 aluminum",
    "density_kg_m3": 2700,

    "yield_strength_MPa": 276,
    "elastic_modulus_GPa": 69,
    "thermal_conductivity_W_mK": 167,
}

# Retrieve values by descriptive keys.
material_name = material_record["material"]
yield_strength = material_record["yield_strength_MPa"]
density = material_record["density_kg_m3"]

# Format a short engineering summary string.
summary = (
    f"{material_name}: yield strength = {yield_strength} MPa, "
    f"density = {density} kg/m^3"
)

# Show selected values from key lookups.
print("Dictionary key lookup example")
print(summary)
print("Available property keys:", list(material_record.keys()))
print("Elastic modulus:", material_record["elastic_modulus_GPa"], "GPa")

```

### **1.2. Meaningful Key Retrieval**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780715746" width="250">



>* Retrieve values using descriptive engineering names
>* Keys clarify each stored property's meaning

>* Clear keys make data easier to maintain
>* Descriptive labels prevent confusing similar values

>* Meaningful keys organize structured engineering records
>* Good keys improve clarity and reuse



```
#@title Python Code - Meaningful Key Retrieval

# Dictionaries retrieve values using descriptive technical labels.
# Meaningful keys reduce confusion in engineering records.
# This example models one pump specification.

# Store pump properties with clear engineering keys.
pump_record = {
    "equipment_id": "P-204",
    "flow_rate_lpm": 320,

    "inlet_pressure_kpa": 185,
    "outlet_pressure_kpa": 460,
    "efficiency_percent": 78,
    "motor_power_kw": 22
}

# Retrieve values by meaning, not position.
flow_rate = pump_record["flow_rate_lpm"]
outlet_pressure = pump_record["outlet_pressure_kpa"]
efficiency = pump_record["efficiency_percent"]

# Compute a simple pressure rise value.
pressure_rise = pump_record["outlet_pressure_kpa"] - pump_record["inlet_pressure_kpa"]

# Print concise results for inspection.
print("Pump ID:", pump_record["equipment_id"])
print("Flow rate (L/min):", flow_rate)
print("Outlet pressure (kPa):", outlet_pressure)
print("Pressure rise (kPa):", pressure_rise)

# Show why a missing key should be handled safely.
maintenance_status = pump_record.get("maintenance_status", "not recorded")
print("Maintenance status:", maintenance_status)

# Format a readable engineering summary string.
summary = f"{pump_record['equipment_id']} runs at {efficiency}% efficiency."
print(summary)

```

### **1.3. Safe Key Access**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780715744" width="250">



>* Handle missing dictionary keys safely
>* Use checks, defaults, and fallback values

>* External data may use inconsistent keys
>* Handle missing values without stopping analysis

>* Judge missing keys by engineering importance
>* Keep workflows robust and data gaps visible



```
#@title Python Code - Safe Key Access

# Safe dictionary access prevents missing-key crashes.
# Engineering records often have incomplete properties.
# Defaults keep analysis clear and robust.

# Store one pump record with missing efficiency.
pump_record = {
    "component_id": "P-204",
    "max_pressure_kpa": 850,

    "temperature_c": 72,
}

# Direct lookup works for existing keys.
pressure_value = pump_record["max_pressure_kpa"]
print("Pressure lookup:", pressure_value, "kPa")

# Safe get returns a clear default.
efficiency_value = pump_record.get("efficiency_percent", "not reported")
print("Efficiency lookup:", efficiency_value)

# Check essential keys before calculations.
required_keys = ["max_pressure_kpa", "temperature_c"]
missing_keys = []

# Record any required property that is absent.
for key_name in required_keys:
    if key_name not in pump_record:
        missing_keys.append(key_name)

# Report whether the record supports review.
if len(missing_keys) == 0:
    print("Required data status: complete")
else:
    print("Missing required keys:", missing_keys)

# Build a clean summary using safe access.
summary_text = (
    "Pump {0}: {1} kPa, {2} C, efficiency {3}"
)

# Format the summary with dictionary values.
formatted_summary = summary_text.format(
    pump_record.get("component_id", "unknown"),
    pump_record.get("max_pressure_kpa", "unknown"),

    pump_record.get("temperature_c", "unknown"),
    pump_record.get("efficiency_percent", "not reported"),
)

# Print the final engineering record summary.
print(formatted_summary)

```

## **2. String Processing**

### **2.1. String Cleaning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780715752" width="250">



>* Prepare messy engineering text for reliable use
>* Standardize case, spaces, and extra characters

>* Remove whitespace for exact text matching
>* Standardize capitalization to avoid duplicate categories

>* Remove interfering characters for predictable processing
>* Clean carefully to preserve engineering meaning



```
#@title Python Code - String Cleaning

# Clean engineering text before structured analysis.
# Small examples make string methods visible.
# Dictionaries store cleaned engineering records.

# Store messy field entries as small strings.
raw_entries = [
    " Pump A  | STEEL | pressure: 1,250 kPa ",
    "\tvalve-02 | aluminum | status = OPEN\n",
    " Sensor 7 | Copper | temperature:  85 C ",
]

# Define a helper for consistent text cleaning.
def clean_text(text_value):
    cleaned_value = text_value.strip().replace("\t", " ")
    cleaned_value = cleaned_value.replace("\n", " ")
    cleaned_value = " ".join(cleaned_value.split())
    return cleaned_value.lower()

# Clean every raw entry before splitting.
cleaned_entries = [clean_text(entry) for entry in raw_entries]
print("Cleaned entries:", cleaned_entries)

# Split cleaned strings into structured pieces.
records = []
for entry in cleaned_entries:
    parts = [piece.strip() for piece in entry.split("|")]
    equipment, material, measurement = parts
    record = {"equipment": equipment, "material": material}

    # Remove labels and commas from measurements.
    measurement = measurement.replace("pressure:", "")
    measurement = measurement.replace("temperature:", "")
    measurement = measurement.replace("status =", "status:")
    measurement = measurement.replace(",", "")

    # Store the cleaned measurement text.
    record["measurement"] = measurement.strip()
    records.append(record)

# Print compact structured engineering records.
print("First record:", records[0])
print("Second record:", records[1])
print("Third record:", records[2])

# Format a clean report line.
summary = f"{records[0]['equipment'].title()} uses {records[0]['material'].title()}."
print("Formatted summary:", summary)

```

### **2.2. String Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780715754" width="250">



>* Makes technical text clear and consistent
>* Turns mixed data into readable records

>* Control numeric precision for readable results
>* Show units and context clearly

>* Consistent records are easier to scan
>* Formatting preserves meaning and reduces confusion



```
#@title Python Code - String Formatting

# Format engineering values for readable reporting.
# Use f-strings with labels and units.
# Combine records into consistent text lines.

# Create one engineering record dictionary.
record = dict(sensor="Pump A", value=12.3456, unit="mm/s")

# Add extra context as formatted strings.
record["status"] = "warning".upper()
record["timestamp"] = "2026-06-06 14:05"

# Build a compact formatted report line.
report = (
    f"{record['timestamp']} | "
    f"{record['sensor']:<8} | "

    f"{record['value']:>6.2f} {record['unit']} | "
    f"{record['status']}"
)

# Format several related records consistently.
readings = [9.8, 12.3456, 15.021]
lines = []

# Convert each numeric reading into text.
for index, reading in enumerate(readings, start=1):
    text = f"Point {index}: {reading:05.2f} mm/s"
    lines.append(text)

# Print only concise teaching output.
print("Single formatted engineering record:")
print(report)
print("Consistent formatted readings:")
print("; ".join(lines))

```

### **2.3. Formatted Strings**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780715756" width="250">



>* Turn raw values into readable messages
>* Add context to reduce technical ambiguity

>* Combine data types with controlled presentation
>* Create consistent, reusable technical output

>* Format technical messages for clear communication
>* Standardize data for accuracy and traceability



```
#@title Python Code - Formatted Strings

# Format measurements into clear engineering messages.
# Combine cleaned strings with rounded values.
# Keep output short and beginner friendly.

# Store raw engineering values in dictionaries.
reading = {
    "equipment": " pump A-17 ",
    "location": " north bay ",
    "pressure_kpa": 486.728,
}

# Clean text values before formatting messages.
equipment = reading["equipment"].strip().upper()
location = reading["location"].strip().title()
pressure = reading["pressure_kpa"]

# Define acceptable engineering pressure limits.
lower_limit = 450.0
upper_limit = 475.0
status = "PASS" if pressure <= upper_limit else "CHECK"

# Build readable labels using formatted strings.
message = (
    f"{equipment} at {location}: "
    f"pressure = {pressure:.1f} kPa, "
    f"limit = {upper_limit:.0f} kPa, "
    f"status = {status}."
)

# Format a compact report line consistently.
report_line = (
    f"Equipment={equipment} | "
    f"Location={location} | "
    f"Pressure={pressure:,.2f} kPa"
)

# Show the formatted engineering messages.
print("Formatted status message:")
print(message)
print("\nStandard report line:")
print(report_line)

```

## **3. Structured Engineering Records**

### **3.1. Creating Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780715759" width="250">



>* Group related engineering details in dictionaries
>* Use meaningful keys for clear retrieval

>* Clean and standardize raw engineering text
>* Store processed strings for reliable records

>* Lists store related repeated engineering values
>* Combined structures create useful evidence records



```
#@title Python Code - Creating Records

# Build one structured engineering record.
# Clean strings before storing values.
# Combine dictionaries, lists, and strings.

# Store raw engineering text values.
raw_tag = "  plant-2 / pump / p-104  "
raw_note = "  vibration within limit; check seal next visit  "
measurements_mm = [12.04, 12.01, 12.03]

# Clean and split the equipment tag.
tag_parts = raw_tag.strip().upper().split("/")
area = tag_parts[0].strip()
system = tag_parts[1].strip()
component = tag_parts[2].strip()

# Clean and standardize the inspection note.
clean_note = raw_note.strip().capitalize()
average_mm = sum(measurements_mm) / len(measurements_mm)
status = "PASS" if average_mm <= 12.05 else "REVIEW"

# Store related fields in one dictionary.
record = {
    "area": area,
    "system": system,
    "component": component,
    "measurements_mm": measurements_mm,
    "average_mm": round(average_mm, 3),
    "status": status,
    "note": clean_note,
}

# Print selected record fields clearly.
print("Structured Engineering Record")
print("Component:", record["component"])
print("Location:", record["area"], record["system"])
print("Measurements mm:", record["measurements_mm"])
print("Average mm:", record["average_mm"])
print("Status:", record["status"])
print("Note:", record["note"])

```

### **3.2. Record Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780715762" width="250">



>* Lists group many structured engineering records
>* Consistent fields make entries comparable

>* Dictionaries name fields; lists collect records
>* Clean strings improve searching and reporting

>* Consistent records enable filtering, sorting, and summaries
>* Record lists support accurate future data use



```
#@title Python Code - Record Lists

# Build structured engineering records.
# Combine dictionaries, lists, and strings.
# Keep records consistent and searchable.

# Raw inspection notes use inconsistent spacing and capitalization.
raw_notes = [
    " pump-01 , north bay , steel , ok ",
    " valve-02, south bay, brass, needs repair ",
    " sensor-03 , roof deck , plastic , ok "
]

# Start an empty list for record dictionaries.
records = []

# Convert each comma-separated note into cleaned fields.
for note in raw_notes:
    parts = [item.strip().title() for item in note.split(",")]
    if len(parts) == 4:
        record = {
            "id": parts[0].upper(),
            "location": parts[1],
            "material": parts[2],
            "status": parts[3]
        }
        records.append(record)

# Print the complete structured record list.
print("Structured record list:")
print(records)

# Filter records using meaningful dictionary keys.
attention = [r for r in records if r["status"] != "Ok"]

# Report one concise engineering summary.
print("Records needing attention:")
print(attention)

# Format one record as a readable maintenance line.
first = records[0]
message = f"{first['id']} at {first['location']} is {first['status']}."
print(message)

```

### **3.3. Record Retrieval**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780715760" width="250">



>* Use keys to find specific record data
>* Organized records make engineering questions easier

>* Filter records using numbers or text
>* Standardize strings to avoid missed matches

>* Retrieved records support analysis and reporting
>* Robust structures handle access, filtering, and sharing



```
#@title Python Code - Record Retrieval

# Retrieve structured engineering records safely.
# Dictionaries hold meaningful engineering fields.
# Lists store related record dictionaries.

# No package installation is required.

# Create small maintenance record dictionaries.
records = [
    dict(tag="P-101", location="Plant A", status="OK"),
    dict(tag="V-204", location="Plant B", status="CHECK"),
    dict(tag="P-102", location="Plant A", status="CHECK"),
]

# Add compact technical notes separately.
records[0]["note"] = "Pump inspected; no corrosion found"
records[1]["note"] = "Valve shows minor corrosion near flange"
records[2]["note"] = "Pump vibration high; schedule alignment"

# Retrieve one record using its tag.
target_tag = "V-204"
matching = [r for r in records if r["tag"] == target_tag]
selected = matching[0] if matching else None

# Clean text for reliable searching.
search_word = "corrosion".lower().strip()
corrosion_records = [
    r for r in records if search_word in r["note"].lower()
]

# Retrieve records from one location.
location_name = "plant a".lower().strip()
plant_a_records = [
    r for r in records if r["location"].lower() == location_name
]

# Format retrieved results for engineers.
print("Selected tag:", selected["tag"], selected["status"])
print("Corrosion notes found:", len(corrosion_records))
print("Plant A records found:", len(plant_a_records))

# Display compact retrieved summaries.
for record in corrosion_records + plant_a_records:
    summary = f"{record['tag']} | {record['location']} | {record['status']}"
    print(summary)

print("Retrieval complete for", len(records), "records")

```
