## **1. Numerical Integration**

### **1.1. Riemann Sums**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1780864882" width="250">



>* Approximate accumulation using many simple rectangles
>* Rectangle width and height estimate total area

>* Narrower rectangles usually give better estimates
>* Sample point choice affects over or underestimates

>* Turn engineering data into accumulated totals
>* Add small measured pieces to approximate effects



```
#@title Python Code - Riemann Sums

# Riemann sums approximate accumulated quantities.
# Rectangles replace curved regions numerically.
# Smaller widths usually improve accuracy.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Define a speed function.
def speed(time_values):
    return 12.0 + 3.0 * time_values

# Set integration interval values.
start_time = 0.0
end_time = 5.0
rectangle_count = 8

# Validate the rectangle count.
if rectangle_count <= 0:
    raise ValueError("rectangle_count must be positive")

# Compute equal rectangle width.
time_width = (end_time - start_time) / rectangle_count
left_times = np.linspace(start_time, end_time - time_width, rectangle_count)
right_times = left_times + time_width

# Compute left and right sums.
left_distance = np.sum(speed(left_times) * time_width)
right_distance = np.sum(speed(right_times) * time_width)

# Compute the exact comparison value.
exact_distance = 12.0 * 5.0 + 1.5 * 5.0**2
absolute_gap = abs(right_distance - left_distance)

# Print a compact engineering summary.
print("Speed model: v(t) = 12 + 3t m/s")
print(f"Rectangles used: {rectangle_count}")
print(f"Left Riemann distance: {left_distance:.2f} m")
print(f"Right Riemann distance: {right_distance:.2f} m")
print(f"Exact distance comparison: {exact_distance:.2f} m")
print(f"Left-right estimate gap: {absolute_gap:.2f} m")

# Prepare smooth curve and rectangles.
smooth_times = np.linspace(start_time, end_time, 200)
smooth_speeds = speed(smooth_times)
rectangle_heights = speed(left_times)

# Create one plot for visualization.
plt.figure(figsize=(7, 4))
plt.plot(smooth_times, smooth_speeds, label="speed curve")
plt.bar(left_times, rectangle_heights, width=time_width, align="edge", alpha=0.35)

# Label the plot clearly.
plt.xlabel("time, t (s)")
plt.ylabel("speed, v(t) (m/s)")
plt.title("Left Riemann Sum for Distance")
plt.legend()

# Display the single figure.
plt.show()

```

### **1.2. Trapezoidal Rule**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1780864884" width="250">



>* Estimate area using trapezoids between data points
>* Assume linear change for better measurements

>* Add many trapezoids for better estimates
>* Convert sampled measurements into accumulated totals

>* Accuracy depends on curve straightness
>* Smaller intervals improve accuracy but cost more



```
#@title Python Code - Trapezoidal Rule

# This script demonstrates trapezoidal numerical integration.
# Speed measurements estimate total traveled distance.
# Results compare against an exact value.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Define measurement times in seconds.
time_seconds = np.linspace(0.0, 10.0, 6)

# Define a known speed function.
speed_mps = 3.0 + 0.4 * time_seconds

# Check the measurement arrays match.
assert time_seconds.shape == speed_mps.shape

# Compute area using trapezoidal rule.
distance_trap = np.trapz(speed_mps, time_seconds)

# Compute exact integral for comparison.
exact_distance = 3.0 * 10.0 + 0.2 * 10.0**2

# Compute the absolute error magnitude.
absolute_error = abs(distance_trap - exact_distance)

# Print a compact numerical summary.
print(f"Trapezoidal estimate: {distance_trap:.2f} meters")
print(f"Exact distance:        {exact_distance:.2f} meters")
print(f"Absolute error:        {absolute_error:.2e} meters")

# Plot samples and trapezoid areas.
plt.figure(figsize=(7, 4))
plt.plot(time_seconds, speed_mps, "o-", label="Speed samples")
plt.fill_between(time_seconds, speed_mps, alpha=0.25, label="Trapezoids")
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (meters per second)")
plt.title("Distance from speed using trapezoidal rule")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

```

### **1.3. Simpson Rule**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_01_03.jpg?v=1780864886" width="250">



>* Fits parabolas through smooth sampled data
>* Estimates areas, totals, and work accurately

>* Curved estimates better model changing speeds
>* Useful for distance, flow, energy, and work

>* Use with smooth, evenly spaced data
>* Check estimates against engineering judgment



```
#@title Python Code - Simpson Rule

# Simpson rule estimates accumulated engineering quantities.
# Smooth sampled curves improve area estimates.
# We compare against an exact distance.

# Import numerical tools and plotting.
import numpy as np

# Import plotting tools for one figure.
import matplotlib.pyplot as plt

# Define evenly spaced time samples.
time = np.linspace(0.0, 10.0, 11)

# Define smooth velocity in meters per second.
velocity = 3.0 + 0.4 * time**2

# Check Simpson rule sample requirement.
point_count = len(time)

# Simpson rule needs an odd point count.
if point_count % 2 == 0:
    raise ValueError("Simpson rule needs odd sampled points")

# Compute the constant time spacing.
dt_values = np.diff(time)

# Verify equal spacing before integration.
if not np.allclose(dt_values, dt_values[0]):
    raise ValueError("Simpson rule needs equal spacing")

# Store the shared sample spacing.
h = dt_values[0]

# Apply Simpson one-third weighted sum.
simpson_distance = (
    h / 3.0 * (
        velocity[0] + velocity[-1]
        + 4.0 * np.sum(velocity[1:-1:2])
        + 2.0 * np.sum(velocity[2:-1:2])
    )
)

# Compute a trapezoid estimate for comparison.
trapezoid_distance = np.trapz(velocity, time)

# Compute exact distance from calculus.
exact_distance = 3.0 * 10.0 + 0.4 * 10.0**3 / 3.0

# Print compact engineering comparison values.
print(f"Simpson distance: {simpson_distance:.3f} m")
print(f"Trapezoid distance: {trapezoid_distance:.3f} m")
print(f"Exact distance: {exact_distance:.3f} m")
print(f"Simpson error: {simpson_distance - exact_distance:.6f} m")

# Create a smooth curve for visualization.
fine_time = np.linspace(0.0, 10.0, 200)

# Evaluate the same velocity model smoothly.
fine_velocity = 3.0 + 0.4 * fine_time**2

# Make exactly one teaching plot.
plt.figure(figsize=(7, 4))

# Plot the velocity curve and samples.
plt.plot(fine_time, fine_velocity, label="smooth velocity")
plt.plot(time, velocity, "o", label="sampled data")

# Shade accumulated distance area under curve.
plt.fill_between(fine_time, fine_velocity, alpha=0.2)

# Label axes and explain the figure.
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.title("Distance equals area under velocity curve")
plt.legend()

# Display the single plot.
plt.show()

```

## **2. Time Stepping Basics**

### **2.1. System State**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1780864888" width="250">



>* State captures key system quantities now
>* Updates predict changes from current state

>* Track enough variables to capture behavior
>* Avoid unnecessary detail that complicates simulation

>* State links intuition to numerical updates
>* Check units, ranges, and physical consistency



```
#@title Python Code - System State

# Track state variables during simulation.
# Update equations predict future states.
# Simple checks reveal unrealistic states.

# Import plotting tools for visualization.
import matplotlib.pyplot as plt

# Store initial state values.
state = {
    "time_s": 0.0,
    "height_m": 1.2,
    "velocity_m_s": 0.0,
}

# Define physical and numerical settings.
gravity_m_s2 = -9.81
time_step_s = 0.05
steps_count = 12

# Create histories for plotting.
time_history = [state["time_s"]]
height_history = [state["height_m"]]
velocity_history = [state["velocity_m_s"]]

# Advance the state repeatedly.
for step_number in range(steps_count):
    state["velocity_m_s"] += gravity_m_s2 * time_step_s
    state["height_m"] += state["velocity_m_s"] * time_step_s
    state["time_s"] += time_step_s

    # Prevent physically impossible negative height.
    if state["height_m"] < 0.0:
        state["height_m"] = 0.0
        state["velocity_m_s"] = 0.0

    # Save each updated state.
    time_history.append(state["time_s"])
    height_history.append(state["height_m"])
    velocity_history.append(state["velocity_m_s"])

# Print a compact state summary.
print("Initial state: height 1.20 m, velocity 0.00 m/s")
print(f"Final time: {state['time_s']:.2f} s")
print(f"Final height: {state['height_m']:.3f} m")
print(f"Final velocity: {state['velocity_m_s']:.3f} m/s")
print(f"Stored state snapshots: {len(time_history)}")

# Plot height history versus time.
plt.figure(figsize=(6, 4))
plt.plot(time_history, height_history, marker="o")
plt.xlabel("time (s)")
plt.ylabel("height (m)")

# Add labels describing the simulation.
plt.title("System state updated through time")
plt.grid(True)
plt.tight_layout()
plt.show()

```

### **2.2. Update Equations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780864889" width="250">



>* Rules estimate the next system state
>* Continuous changes become step-by-step calculations

>* Assume rates stay constant over small steps
>* Convert physical laws into stepwise predictions

>* Update rules must match real behavior
>* Small time steps reduce unstable errors



```
#@title Python Code - Update Equations

# Update equations advance simulated systems.
# Small time steps improve predictions.
# Plots reveal dynamic behavior clearly.

import math
import numpy as np
import matplotlib.pyplot as plt

# Define a cooling model.
room_temperature = 22.0
initial_temperature = 90.0
cooling_constant = 0.18

# Choose two time steps.
coarse_dt = 1.0
fine_dt = 0.1
total_time = 20.0

# Simulate with explicit Euler updates.
def simulate_cooling(time_step):
    steps = int(total_time / time_step)
    times = np.linspace(0.0, total_time, steps + 1)

    temperatures = np.zeros(steps + 1)
    temperatures[0] = initial_temperature

    for index in range(steps):
        current = temperatures[index]
        rate = -cooling_constant * (current - room_temperature)

        temperatures[index + 1] = current + rate * time_step
    return times, temperatures

# Compute coarse and fine solutions.
coarse_time, coarse_temp = simulate_cooling(coarse_dt)
fine_time, fine_temp = simulate_cooling(fine_dt)

# Compute the exact reference curve.
exact_temp = room_temperature + (initial_temperature - room_temperature) * np.exp(-cooling_constant * fine_time)
final_error = abs(fine_temp[-1] - exact_temp[-1])

# Print a compact summary.
print("Update rule: next T = current T + rate * dt")
print(f"Fine final temperature: {fine_temp[-1]:.2f} deg C")
print(f"Exact final temperature: {exact_temp[-1]:.2f} deg C")
print(f"Fine final error: {final_error:.3f} deg C")
print("Smaller dt follows the exact curve more closely.")

# Plot the simulated temperatures.
plt.figure(figsize=(7, 4))
plt.plot(coarse_time, coarse_temp, "o-", label="dt = 1.0 min")
plt.plot(fine_time, fine_temp, "-", label="dt = 0.1 min")
plt.plot(fine_time, exact_temp, "--", label="exact reference")
plt.xlabel("time, minutes")

# Add labels and display.
plt.ylabel("temperature, deg C")
plt.title("Cooling simulation using update equations")
plt.grid(True)
plt.legend()
plt.show()

```

### **2.3. State Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1780864891" width="250">



>* Advance the system one time step
>* Build simulations through repeated updates

>* Time step size affects simulation accuracy.
>* Balance detail with computational cost.

>* Carry forward all important changing variables
>* Create a consistent state for next step



```
#@title Python Code - State Updates

# State updates advance simulations one step.
# Small time steps improve dynamic tracking.
# This example models a falling ball.

import math
import matplotlib.pyplot as plt

# Define gravity and the time step.
gravity = 9.81
step_size = 0.05

# Define the starting state values.
time_value = 0.0
height_value = 20.0
velocity_value = 0.0

# Store state history for plotting.
time_history = [time_value]
height_history = [height_value]
velocity_history = [velocity_value]

# Set a safe upper step count.
maximum_steps = 500
step_index = 0

# Advance until the ball reaches ground.
while height_value > 0.0 and step_index < maximum_steps:
    acceleration_value = -gravity
    velocity_value = velocity_value + acceleration_value * step_size

    height_value = height_value + velocity_value * step_size
    time_value = time_value + step_size
    step_index = step_index + 1

    time_history.append(time_value)
    height_history.append(max(height_value, 0.0))
    velocity_history.append(velocity_value)

# Compare numerical impact time with exact value.
exact_time = math.sqrt(2.0 * 20.0 / gravity)
numerical_time = time_history[-1]
time_error = numerical_time - exact_time

# Print a concise simulation summary.
print(f"Initial height: {20.0:.1f} m")
print(f"Time step: {step_size:.2f} s")
print(f"Numerical impact time: {numerical_time:.2f} s")
print(f"Exact impact time: {exact_time:.2f} s")
print(f"Impact time error: {time_error:.3f} s")
print(f"Recorded states: {len(time_history)} snapshots")

# Plot height values against time.
plt.figure(figsize=(7, 4))
plt.plot(time_history, height_history, marker="o", markersize=3)
plt.xlabel("time (s)")
plt.ylabel("height (m)")
plt.title("State updates for falling-ball height")
plt.grid(True)
plt.show()

```

## **3. Result Checking**

### **3.1. Known Value Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780864877" width="250">



>* Test methods on problems with known answers
>* Good checks increase confidence in complex results

>* Known checks reveal hidden numerical mistakes
>* Simple benchmarks separate errors from real behavior

>* Choose checks matching real problem behavior
>* Use step tests to build trust



```
#@title Python Code - Known Value Checks

# Known values help test numerical methods.
# Simple cases can reveal hidden mistakes.
# Smaller steps should improve trusted comparisons.

import math
import numpy as np
import matplotlib.pyplot as plt

# Define a curve with known area.
start_x, end_x = 0.0, math.pi
known_area = 2.0
step_counts = [4, 8, 16, 32]

# Store numerical estimates for comparison.
estimates = []
errors = []

# Apply the trapezoid rule with different steps.
for intervals in step_counts:
    x_values = np.linspace(start_x, end_x, intervals + 1)
    y_values = np.sin(x_values)
    estimate = np.trapz(y_values, x_values)

    error = abs(estimate - known_area)
    estimates.append(estimate)
    errors.append(error)

# Print a compact known value check table.
print("Known value check: area under sin(x) from 0 to pi")
print(f"Exact area = {known_area:.6f} square meters")
print("Intervals | Estimate | Absolute error")
for intervals, estimate, error in zip(step_counts, estimates, errors):
    print(f"{intervals:9d} | {estimate:8.6f} | {error:14.6f}")

# Check whether accuracy improves as steps shrink.
improves = all(errors[i] > errors[i + 1] for i in range(len(errors) - 1))
print(f"Errors decrease with smaller steps: {improves}")

# Plot error trend on one clear figure.
plt.figure(figsize=(6, 4))
plt.plot(step_counts, errors, marker="o")
plt.xlabel("Number of trapezoids")
plt.ylabel("Absolute error")
plt.title("Known Value Check for Numerical Integration")
plt.grid(True)
plt.show()

```

### **3.2. Known Value Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780864881" width="250">



>* Compare results with simple known cases
>* Catch setup, unit, and logic errors

>* Choose simple benchmarks resembling the real problem
>* Large mismatches reveal setup or method errors

>* Start simple, then add complexity gradually
>* Check trusted values to avoid false confidence



```
#@title Python Code - Known Value Checks

# Known values make numerical checks trustworthy.
# Simple benchmarks reveal setup mistakes early.
# This example checks work integration.

import numpy as np
import matplotlib.pyplot as plt

# Define a constant force benchmark.
force_newtons = 50.0

# Define the travel distance in meters.
distance_meters = 4.0

# Choose small step counts for comparison.
step_counts = [4, 20, 100]

# Compute the exact work value.
exact_work = force_newtons * distance_meters

# Store numerical work estimates here.
numerical_results = []

# Test the same known case repeatedly.
for count in step_counts:
    positions = np.linspace(0.0, distance_meters, count + 1)
    forces = np.full_like(positions, force_newtons)
    work = np.trapezoid(forces, positions)

    error = abs(work - exact_work)
    numerical_results.append(work)
    print(f"steps={count:3d}, work={work:7.2f} J, error={error:.2e}")

# Print the known benchmark answer.
print(f"known value = {exact_work:.2f} J")

# Plot estimates against the benchmark.
plt.figure(figsize=(6, 4))
plt.plot(step_counts, numerical_results, marker="o", label="numerical")
plt.axhline(exact_work, color="red", linestyle="--", label="known value")

# Label the plot for interpretation.
plt.xlabel("number of integration steps")
plt.ylabel("work estimate, joules")
plt.title("Known Value Check: Constant Force Work")
plt.legend()

# Show the single teaching plot.
plt.show()

```

### **3.3. Conservation Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780864879" width="250">



>* Check simulations against physical conservation laws
>* Balance violations reveal numerical or setup problems

>* Track conserved quantities against expected physical changes
>* Use mismatches to spot numerical errors

>* Find hidden modeling and algorithm problems
>* Check changes match model assumptions



```
#@title Python Code - Conservation Checks

# Conservation checks catch hidden simulation errors.
# This script audits a simple tank model.
# Stored volume should match accumulated flow.

import numpy as np
import matplotlib.pyplot as plt

# Define a small deterministic time grid.
time_step = 0.5
end_time = 10.0
steps = int(end_time / time_step) + 1

# Validate the chosen grid size.
if steps < 2:
    raise ValueError("Need at least two time points")

# Create times and flow rates.
time = np.linspace(0.0, end_time, steps)
inflow = 2.0 + 0.3 * np.sin(time)
outflow = 1.2 + 0.1 * np.cos(time)

# Allocate tank volume history.
volume = np.zeros(steps)
volume[0] = 5.0
net_flow = inflow - outflow

# Update storage using Euler time stepping.
for index in range(1, steps):
    volume[index] = volume[index - 1] + net_flow[index - 1] * time_step

# Compute independent conservation balance.
net_added = np.sum(net_flow[:-1] * time_step)
predicted_final = volume[0] + net_added
balance_error = volume[-1] - predicted_final

# Summarize the physical audit.
print("Tank conservation check")
print(f"Initial volume: {volume[0]:.3f} m^3")
print(f"Final volume: {volume[-1]:.3f} m^3")
print(f"Accumulated net inflow: {net_added:.3f} m^3")
print(f"Balance error: {balance_error:.3e} m^3")

# Judge whether bookkeeping is consistent.
tolerance = 1e-12
passed_check = abs(balance_error) < tolerance
print(f"Conservation check passed: {passed_check}")

# Plot storage and balance prediction.
plt.figure(figsize=(7, 4))
plt.plot(time, volume, marker="o", label="Simulated storage")
plt.axhline(predicted_final, linestyle="--", label="Final balance prediction")
plt.xlabel("Time (s)")
plt.ylabel("Tank volume (m^3)")
plt.title("Conservation Check for Tank Storage")
plt.legend()
plt.grid(True)
plt.show()

```
