## **1. Dictionary Basics**

### **1.1. Mapping Engineering Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1780894798" width="250">



>* Dictionaries connect engineering labels to data
>* Keys make project information easier to find

>* Labels connect identifiers to engineering information
>* Dictionaries improve readability and project efficiency

>* Unique labels prevent ambiguous dictionary mappings
>* Values can store simple or detailed records



```
#@title Python Code - Mapping Engineering Labels

# Dictionaries connect labels to engineering information.
# Keys should be unique project identifiers.
# Values can store properties or records.

# Create a dictionary mapping labels to beam lengths.
beam_lengths_m = {
    "B1": 6.0,
    "B2": 7.5,

    "B3": 5.25,
}

# Access one value using its engineering label.
selected_length = beam_lengths_m["B2"]
print("B2 length in meters:", selected_length)

# Create richer records using nested dictionaries.
member_records = {
    "B1": {"material": "steel", "length_m": 6.0},
    "B2": {"material": "concrete", "length_m": 7.5},

    "B3": {"material": "timber", "length_m": 5.25},
}

# Read one property from one labeled record.
material_b1 = member_records["B1"]["material"]
print("B1 material:", material_b1)

# Add a new labeled member record.
member_records["B4"] = {"material": "steel", "length_m": 4.8}
print("Number of member records:", len(member_records))

# Update an existing property by label.
member_records["B3"]["length_m"] = 5.40
updated_length = member_records["B3"]["length_m"]

print("Updated B3 length in meters:", updated_length)

# List labels to show the dictionary mapping.
labels = list(member_records.keys())
print("Member labels:", labels)

```

### **1.2. Mapping Engineering Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1780894801" width="250">



>* Dictionaries link labels to related information
>* Use named engineering objects, not positions

>* Labels act as unique dictionary keys
>* Values store measurements, properties, or records

>* Choose stable, clear engineering labels
>* Good labels reduce errors and confusion



```
#@title Python Code - Mapping Engineering Labels

# Dictionaries connect labels to engineering information.
# Stable labels make project data easier.
# This example maps beams to materials.

# Create a small dictionary of beam materials.
beam_materials = {
    "B1": "A992 structural steel",
    "B2": "reinforced concrete",

    "B3": "glulam timber",
}

# Access values using meaningful engineering labels.
selected_beam = "B2"
selected_material = beam_materials[selected_beam]

# Add another labeled engineering component.
beam_materials["B4"] = "precast concrete"

# Update a value when project information changes.
beam_materials["B3"] = "treated glulam timber"

# Store a richer record for one foundation.
foundation_record = {
    "type": "spread footing",
    "width_m": 2.4,

    "depth_ft": 5.0,
}

# Print concise results for the learner.
print("Beam labels:", list(beam_materials.keys()))
print(selected_beam, "material:", selected_material)
print("Updated B3:", beam_materials["B3"])
print("Foundation type:", foundation_record["type"])
print("Foundation width, m:", foundation_record["width_m"])

```

### **1.3. Engineering Property Maps**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1780894799" width="250">



>* Labels act as keys for engineering values
>* Property names make data easy to find

>* Map component labels to detailed records
>* Track named parts and their properties

>* Clear labels make data easier to manage
>* Property maps preserve engineering context



```
#@title Python Code - Engineering Property Maps

# Dictionaries connect labels with engineering properties.
# Property maps make project data readable.
# Keys retrieve values without position counting.

# Create a simple material property map.
material_density = {
    "steel": 7850,
    "aluminum": 2700,
}

# Access values using meaningful labels.
steel_density = material_density["steel"]
aluminum_density = material_density["aluminum"]
print("Steel density:", steel_density, "kg/m^3")

# Add another material to the map.
material_density["concrete"] = 2400
print("Concrete density:", material_density["concrete"], "kg/m^3")

# Create records for named structural beams.
beam_records = {
    "B1": {"span_ft": 24, "section": "W12x26"},
    "B2": {"span_ft": 18, "section": "W10x22"},
}

# Retrieve one beam record directly.
selected_beam = beam_records["B1"]
print("Beam B1 record:", selected_beam)

# Update one property inside a record.
beam_records["B2"]["span_ft"] = 20
print("Updated B2 span:", beam_records["B2"]["span_ft"], "ft")

# Show available engineering labels.
print("Material labels:", list(material_density.keys()))
print("Beam labels:", list(beam_records.keys()))

```

## **2. Updating Project Records**

### **2.1. Accessing Record Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1780894794" width="250">



>* Dictionaries store project records using named fields
>* Labels make values easier to access

>* Check whether requested fields exist
>* Handle missing, empty, and valid values

>* Retrieve fields to support project decisions
>* Use clear labels for auditable workflows



```
#@title Python Code - Accessing Record Values

# Dictionary records store engineering project fields.
# Field labels make values easier to access.
# Safe access prevents missing-data interruptions.

# Create one bridge component record.
component_record = {
    "component_id": "BRG-17",
    "material": "reinforced concrete",
    "span_m": 24.5,

    "inspection_status": "needs follow-up",
    "load_rating_tons": 38,
    "last_inspection": "2026-04-12",
}

# Access values using exact dictionary keys.
component_id = component_record["component_id"]
material_name = component_record["material"]
status_text = component_record["inspection_status"]

# Use accessed values in formatted messages.
print(f"Component: {component_id}")
print(f"Material: {material_name}")
print(f"Status: {status_text}")

# Access numeric fields for calculations.
span_meters = component_record["span_m"]
span_feet = span_meters * 3.28084
print(f"Span: {span_meters} m ({span_feet:.1f} ft)")

# Use get for optional record fields.
repair_priority = component_record.get("repair_priority", "not assigned")
print(f"Repair priority: {repair_priority}")

# Check whether a field exists.
has_load_rating = "load_rating_tons" in component_record
print(f"Load rating available: {has_load_rating}")

# Access a value only after checking.
if has_load_rating:
    rating_value = component_record["load_rating_tons"]
    print(f"Load rating: {rating_value} tons")

# Summarize available record labels.
field_count = len(component_record)
print(f"Recorded fields: {field_count}")

```

### **2.2. Updating Field Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1780894796" width="250">



>* Update dictionary fields as project details change
>* Keep field labels while replacing values

>* Update fields to keep records accurate
>* Use field names to preserve context

>* Update fields accurately and consistently
>* Preserve history when changes affect traceability



```
#@title Python Code - Updating Field Values

# Dictionaries store editable project records.
# Field labels identify current values.
# Updates replace values, not labels.

# Create one bridge inspection record.
bridge_record = {
    "bridge_id": "BR-104",
    "condition_rating": 7,

    "last_inspection": "2026-01-15",
    "recommended_action": "routine monitoring",
    "repair_cost_usd": 12000,
}

# Show selected fields before updating.
print("Before update:", bridge_record["condition_rating"], bridge_record["recommended_action"])

# Update existing field values after inspection.
bridge_record["condition_rating"] = 5
bridge_record["last_inspection"] = "2026-06-08"

# Update action and cost consistently.
bridge_record["recommended_action"] = "schedule deck repairs"
bridge_record["repair_cost_usd"] = 28500

# Store a metric field update too.
bridge_record["crack_width_mm"] = 3.2

# Show selected fields after updating.
print("After update:", bridge_record["condition_rating"], bridge_record["recommended_action"])

# Print the revised record compactly.
print("Inspection date:", bridge_record["last_inspection"])
print("Estimated repair cost:", bridge_record["repair_cost_usd"], "USD")
print("Crack width:", bridge_record["crack_width_mm"], "mm")

```

### **2.3. Updating Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1780894803" width="250">



>* Dictionaries store changing project record details
>* Updates keep datasets current without rebuilding

>* Revise fields or add new project details
>* Update carefully to avoid data errors

>* Keep record updates accurate and consistent
>* Make deliberate changes that support decisions



```
#@title Python Code - Updating Records

# Update engineering records using dictionaries.
# Each record stores changing project information.
# Careful updates keep datasets reliable.

# Create one small beam inspection record.
beam_record = {
    "beam_id": "B-12",
    "location": "Level 2",

    "span_ft": 24.0,
    "status": "Awaiting inspection",
    "engineer": "A. Patel",
    "load_rating_kips": None,

    "inspection_date": None,
}

# Print selected values before updates.
print("Before:", beam_record["status"])
print("Engineer:", beam_record["engineer"])

# Update existing fields after inspection.
beam_record["status"] = "Inspection complete"
beam_record["load_rating_kips"] = 82.5
beam_record["inspection_date"] = "2026-06-08"
beam_record["engineer"] = "M. Rivera"

# Add new fields as information grows.
beam_record["deflection_in"] = 0.31
beam_record["approval"] = "Approved for next phase"

# Remove a temporary field safely.
removed_value = beam_record.pop("temporary_note", "No temporary note")

# Print selected values after updates.
print("After:", beam_record["status"])
print("Rating kips:", beam_record["load_rating_kips"])
print("Deflection inches:", beam_record["deflection_in"])

# Show safe removal and final keys.
print("Removed:", removed_value)
print("Final fields:", sorted(beam_record.keys()))

```

## **3. Set Operations**

### **3.1. Unique Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_03_01.jpg?v=1780894804" width="250">



>* Sets keep one copy of each value
>* Use sets to find distinct engineering data

>* Explore unique engineering data values early
>* Check expected identifiers and unexpected labels

>* Sets show membership, not order or frequency
>* Use sets first for distinct categories



```
#@title Python Code - Unique Values

# Sets keep only unique engineering labels.
# Repeated records often hide distinct categories.
# This example inspects pavement survey data.

# Store repeated pavement classifications from inspections.
pavement_classes = [
    "asphalt", "concrete", "asphalt",
    "gravel", "concrete", "asphalt",
]

# Convert the list into a set.
unique_classes = set(pavement_classes)

# Sort results for beginner friendly display.
sorted_classes = sorted(unique_classes)

# Compare original length against unique length.
record_count = len(pavement_classes)
unique_count = len(unique_classes)

# Detect whether repeated values were present.
has_duplicates = record_count != unique_count

# Print a compact inspection summary.
print("All recorded pavement classes:", pavement_classes)
print("Unique pavement classes:", sorted_classes)
print("Number of survey records:", record_count)
print("Number of unique classes:", unique_count)
print("Were duplicates found?", has_duplicates)

```

### **3.2. Duplicate Detection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1780894806" width="250">



>* Sets reveal repeated engineering identifiers
>* Duplicates support early data quality checks

>* Track seen values to flag repeats
>* Fast checks reveal risky duplicate identifiers

>* Unique sets can hide repeated entries
>* Track duplicates and judge their meaning



```
#@title Python Code - Duplicate Detection

# Sets help detect repeated engineering identifiers.
# Duplicate labels can create project confusion.
# This example checks pile location IDs.

# Store incoming pile identifiers from field records.
pile_ids = [
    "P-101", "P-102", "P-103", "P-102",
    "P-104", "P-105", "P-103", "P-106",
]

# Validate the sample before checking duplicates.
record_count = len(pile_ids)
unique_raw_count = len(set(pile_ids))
assert record_count >= unique_raw_count >= 1

# Create sets for seen values and duplicates.
seen_ids = set()
duplicate_ids = set()
first_time_ids = []

# Scan each identifier once from the list.
for pile_id in pile_ids:
    if pile_id in seen_ids:
        duplicate_ids.add(pile_id)

    else:
        seen_ids.add(pile_id)
        first_time_ids.append(pile_id)

# Summarize unique and duplicate identifier checks.
print("Total field records:", record_count)
print("Unique pile IDs:", len(seen_ids))
print("Duplicate pile IDs:", sorted(duplicate_ids))

# Show first occurrences preserved in project order.
print("First-time pile IDs:", first_time_ids)
print("Duplicates need review:", len(duplicate_ids) > 0)

```

### **3.3. Common Elements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1780894808" width="250">



>* Find shared values across engineering datasets
>* Quickly confirm overlap between data sources

>* Compare team datasets to find shared elements
>* Use overlaps to prioritize engineering decisions

>* Reveal hidden overlaps in messy engineering data
>* Check membership for shared identifiers efficiently



```
#@title Python Code - Common Elements

# Sets find shared engineering identifiers.
# Common elements reveal project overlap.
# This example compares inspection lists.

# Store inspected bridge component identifiers.
field_survey_components = {
    "B01",
    "B02",
    "B03",
    "B05",
}

# Store scheduled maintenance component identifiers.
maintenance_schedule_components = {
    "B02",
    "B04",
    "B05",
    "B06",
}

# Use intersection to find shared identifiers.
common_components = (
    field_survey_components
    & maintenance_schedule_components
)

# Sort results for stable beginner-friendly output.
sorted_common_components = sorted(
    common_components
)

# Count each compared set safely.
survey_count = len(field_survey_components)
schedule_count = len(maintenance_schedule_components)
common_count = len(sorted_common_components)

# Print a compact engineering interpretation.
print("Field survey unique components:", survey_count)
print("Maintenance schedule unique components:", schedule_count)
print("Common components needing attention:", sorted_common_components)
print("Number of common components:", common_count)

```
