## **1. List Storage**

### **1.1. Creating Measurement Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780893927" width="250">



>* Ordered measurements belong together in lists
>* Lists model clear engineering records

>* Use consistent units and measurement types
>* Organize lists for clear comparison

>* Define what list order represents
>* Preserve sequence for meaningful retrieval



```
#@title Python Code - Creating Measurement Lists

# Lists preserve ordered engineering measurements.
# Each value shares one unit.
# Positions help retrieve recorded observations.

# Store weekly settlement readings in millimetres.
settlement_mm = [0.0, 2.4, 4.1, 5.3]

# Store slump measurements in inches.
slump_inches = [3.5, 4.0, 3.75, 4.25]

# Count readings before retrieving positions.
settlement_count = len(settlement_mm)

# Retrieve first and latest settlement readings.
first_settlement = settlement_mm[0]
latest_settlement = settlement_mm[settlement_count - 1]

# Retrieve the second recorded slump measurement.
second_slump = slump_inches[1]

# Print compact field log summaries.
print("Settlement readings, millimetres:", settlement_mm)
print("Number of settlement readings:", settlement_count)
print("First settlement reading:", first_settlement, "mm")
print("Latest settlement reading:", latest_settlement, "mm")
print("Slump readings, inches:", slump_inches)
print("Second slump reading:", second_slump, "in")

```

### **1.2. List Indexing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780893929" width="250">



>* Retrieve measurements by their list position
>* List order preserves engineering sequence meaning

>* List positions start at zero
>* Correct indexes prevent nearby measurement mistakes

>* Clear list order supports quality control
>* Indexes link measurements to real locations



```
#@title Python Code - List Indexing

# Lists store measurements in field order.
# Indexing retrieves one stored measurement.
# Python counts list positions from zero.

# Store pavement thickness measurements in inches.
thickness_inches = [8.1, 7.9, 8.3, 8.0, 7.8]
locations = ["left shoulder", "left wheel path", "centerline", "right wheel path", "right shoulder"]

# Confirm both lists describe matching positions.
if len(thickness_inches) != len(locations):
    raise ValueError("Measurement and location lists must match.")

# Retrieve values using zero based indexes.
first_index = 0
centerline_index = 2
last_index = len(thickness_inches) - 1

# Connect each index to field meaning.
first_label = locations[first_index]
centerline_label = locations[centerline_index]
last_label = locations[last_index]

# Get the stored measurements at those indexes.
first_value = thickness_inches[first_index]
centerline_value = thickness_inches[centerline_index]
last_value = thickness_inches[last_index]

# Print a compact indexing summary.
print("Pavement thickness list:", thickness_inches)
print("Index", first_index, "is", first_label, "=", first_value, "in")
print("Index", centerline_index, "is", centerline_label, "=", centerline_value, "in")
print("Index", last_index, "is", last_label, "=", last_value, "in")

# Show the common zero indexing idea.
physical_position = centerline_index + 1
print("Centerline is physical position", physical_position, "but index", centerline_index)

# Compute one simple engineering check.
difference_inches = centerline_value - first_value
print("Centerline minus first thickness =", difference_inches, "in")

```

### **1.3. Storing Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780893930" width="250">



>* Lists keep measurements in collection order
>* Order supports clear, reliable later analysis

>* List positions give measurements engineering meaning
>* Enter values in the intended sequence

>* Keep units, precision, and meaning consistent
>* Document order for retrieval and calculations



```
#@title Python Code - Storing Measurements

# Lists preserve ordered engineering measurements.
# Positions connect values to field locations.
# Consistent units make retrieval reliable.

# Store rut depths in survey order.
rut_depths_mm = [4.2, 5.1, 6.0, 5.6]

# Store matching segment names separately.
segments = ["A", "B", "C", "D"]

# Confirm both lists describe same route.
if len(rut_depths_mm) != len(segments):
    raise ValueError("Segments and measurements must match.")

# Retrieve measurements using positions.
first_rut = rut_depths_mm[0]
third_rut = rut_depths_mm[2]

# Show how order gives meaning.
print("Rut depths in millimeters:", rut_depths_mm)
print("First segment", segments[0], "rut depth:", first_rut, "mm")
print("Third segment", segments[2], "rut depth:", third_rut, "mm")

# Add one new measurement last.
rut_depths_mm.append(6.4)
segments.append("E")

# Retrieve the newest recorded value.
last_position = len(rut_depths_mm) - 1
last_segment = segments[last_position]
last_rut = rut_depths_mm[last_position]

# Summarize stored ordered data.
print("Added segment", last_segment, "rut depth:", last_rut, "mm")
print("Total stored measurements:", len(rut_depths_mm))
print("Survey order remains:", segments)

```

## **2. Tuple Records**

### **2.1. Coordinate Tuples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780893938" width="250">



>* Tuples group coordinate values as one point
>* Value order defines each coordinate’s meaning

>* Tuples keep coordinate records stable
>* Corrections create new coordinate records

>* Store each location as one tuple
>* Organize many points with clear structure



```
#@title Python Code - Coordinate Tuples

# Coordinate tuples group related location values.
# Tuple order gives engineering meaning.
# Fixed records reduce accidental coordinate changes.

# Store fixed two-dimensional survey coordinates.
benchmark_a = (125.4, 82.7)
column_b3 = (140.0, 96.5)

# Retrieve values by their tuple positions.
east_a, north_a = benchmark_a
east_b, north_b = column_b3

# Compute coordinate changes between two points.
delta_east = east_b - east_a
delta_north = north_b - north_a

# Store a three-dimensional point with elevation.
inlet_point = (118.2, 74.9, 6.3)
inlet_east, inlet_north, inlet_elev = inlet_point

# Keep many fixed coordinate records together.
site_points = [benchmark_a, column_b3, inlet_point[:2]]
point_count = len(site_points)

# Display concise engineering coordinate results.
print("Benchmark A tuple:", benchmark_a)
print("Column B3 tuple:", column_b3)
print("East change in meters:", round(delta_east, 2))
print("North change in meters:", round(delta_north, 2))
print("Inlet elevation in meters:", inlet_elev)
print("Number of plan points:", point_count)

```

### **2.2. Tuple Storage**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1780893934" width="250">



>* Tuples group related measurements in fixed order
>* Value positions must follow shared meaning

>* Tuples store fixed engineering records
>* Stable records reduce workflow mistakes

>* Tuples keep each engineering record organized
>* Collections manage many fixed records consistently



```
#@title Python Code - Tuple Storage

# Tuples store fixed engineering records.
# Positions inside each tuple have meaning.
# Lists can hold many tuple records.

# Store borehole coordinates as fixed pairs.
borehole_a = (12.5, 30.0)
borehole_b = (18.0, 42.5)
borehole_c = (25.5, 37.0)

# Combine fixed records in a list.
borehole_locations = [borehole_a, borehole_b, borehole_c]
record_count = len(borehole_locations)

# Read values by agreed positions.
first_easting = borehole_locations[0][0]
first_northing = borehole_locations[0][1]

# Store beam dimensions as fixed records.
beam_size_in = (12.0, 24.0)
plate_size_mm = (300.0, 200.0, 12.0)

# Display compact tuple records clearly.
print("Borehole coordinate tuples:", borehole_locations)
print("Number of borehole records:", record_count)
print("First easting in meters:", first_easting)
print("First northing in meters:", first_northing)
print("Beam size, width then depth in inches:", beam_size_in)
print("Plate size, length width thickness in mm:", plate_size_mm)

```

### **2.3. Fixed Dimension Records**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780893936" width="250">



>* Tuples group related dimensions in order
>* Fixed positions give records clear meaning

>* Tuples keep related dimensions together
>* Fixed records prevent confusing extra values

>* Tuples make dimension records consistent and comparable
>* Fixed structure reduces ambiguity in engineering data



```
#@title Python Code - Fixed Dimension Records

# Tuples store fixed engineering dimension records.
# Positions give each dimension its meaning.
# Lists can hold many tuple records.

# Create bearing pad records as tuples.
pad_a = (600, 400, 50)
pad_b = (600, 400, 65)
pad_c = (750, 450, 50)

# Store the fixed records in one list.
bearing_pads = [pad_a, pad_b, pad_c]
record_count = len(bearing_pads)
print("Number of bearing pad records:", record_count)

# Retrieve dimensions by agreed positions.
length_mm, width_mm, thickness_mm = bearing_pads[0]
print("First pad length, width, thickness:", length_mm, width_mm, thickness_mm)

# Compare fixed footprint dimensions only.
same_footprint = bearing_pads[0][:2] == bearing_pads[1][:2]
print("First two pads share footprint:", same_footprint)

# Compute footprint areas from tuple positions.
areas_mm2 = [pad[0] * pad[1] for pad in bearing_pads]
average_area = sum(areas_mm2) / len(areas_mm2)
print("Average footprint area in mm^2:", average_area)

# Convert the first tuple record to meters.
first_pad_m = tuple(value / 1000 for value in bearing_pads[0])
print("First pad dimensions in meters:", first_pad_m)

```

## **3. Processing List Data**

### **3.1. Summarizing Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780893939" width="250">



>* Lists organize ordered engineering measurements
>* Summaries reveal totals, changes, and patterns

>* Compare ordered measurements to reveal changes
>* Interpret unusual values using nearby readings

>* Choose summaries based on measurement meaning
>* Clear summaries support analysis and communication



```
#@title Python Code - Summarizing Measurements

# Lists help summarize repeated engineering measurements.
# Ordered values support totals and differences.
# Small summaries reveal practical field patterns.

# Store rainfall depths from consecutive monitoring days.
rainfall_mm = [12.5, 8.0, 0.0, 15.5, 6.0]

# Validate that measurements contain usable numeric values.
measurement_count = len(rainfall_mm)

# Compute total and average rainfall from the list.
total_rainfall_mm = sum(rainfall_mm)
average_rainfall_mm = total_rainfall_mm / measurement_count

# Compare final and first readings for change.
first_depth_mm = rainfall_mm[0]
last_depth_mm = rainfall_mm[-1]
net_change_mm = last_depth_mm - first_depth_mm

# Create neighboring differences across the ordered list.
daily_changes_mm = []

# Process each adjacent pair safely.
for index in range(1, measurement_count):
    change_mm = rainfall_mm[index] - rainfall_mm[index - 1]
    daily_changes_mm.append(change_mm)

# Print a concise engineering summary.
print("Rainfall measurements, mm:", rainfall_mm)
print("Number of observations:", measurement_count)
print("Total rainfall, mm:", total_rainfall_mm)
print("Average rainfall, mm:", average_rainfall_mm)

# Print ordered change information for interpretation.
print("Net change from first to last, mm:", net_change_mm)
print("Daily changes, mm:", daily_changes_mm)

```

### **3.2. Summarizing List Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780893941" width="250">



>* Turn ordered measurements into useful summaries
>* Support engineering decisions and quality control

>* Add list values to find accumulated totals
>* Compare ordered values to reveal changes

>* Match summaries to measurement meaning
>* Check spacing, gaps, outliers, and units



```
#@title Python Code - Summarizing List Values

# Summaries turn measurements into useful engineering information.
# Lists keep ordered field readings together clearly.
# Python can total, average, and compare values.

# Store daily concrete deliveries in cubic yards.
truck_volumes_cy = [7.5, 8.0, 7.25, 8.5, 7.75]

# Count readings before computing the average.
reading_count = len(truck_volumes_cy)

# Validate the list before summary calculations.
if reading_count > 0:

    # Add all truck volumes for total concrete.
    total_volume_cy = sum(truck_volumes_cy)

    # Divide total volume by number of readings.
    average_volume_cy = total_volume_cy / reading_count

    # Compare each reading with the previous reading.
    volume_changes_cy = [
        truck_volumes_cy[index] - truck_volumes_cy[index - 1]
        for index in range(1, reading_count)
    ]

    # Print concise summary results for learners.
    print("Truck volumes, cubic yards:", truck_volumes_cy)
    print("Total delivered, cubic yards:", total_volume_cy)
    print("Average truck load, cubic yards:", round(average_volume_cy, 2))
    print("Changes between trucks, cubic yards:", volume_changes_cy)

    # Summarize ordered survey elevations in meters.
    elevations_m = [101.2, 101.6, 101.1, 100.8, 101.0]

    # Compute station-to-station elevation differences.
    elevation_changes_m = [
        elevations_m[index] - elevations_m[index - 1]
        for index in range(1, len(elevations_m))
    ]

    # Print elevation summary without excessive output.
    print("Elevation changes, meters:", elevation_changes_m)
    print("Average elevation, meters:", round(sum(elevations_m) / len(elevations_m), 2))
else:

    # Explain the safe path for empty lists.
    print("No readings are available to summarize.")

```

### **3.3. Averaging Measurements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780893943" width="250">



>* Average repeated measurements to find typical values
>* Use averages carefully with unusual readings

>* Average equals total divided by measurement count
>* Meaning depends on units and sampling

>* Averages can hide critical measurement problems
>* Compare with ranges, limits, and full data



```
#@title Python Code - Averaging Measurements

# Lists store repeated engineering measurements.
# Averages summarize typical measurement levels.
# Individual values still remain important.

# Store slab thickness readings in inches.
thickness_inches = [5.9, 6.1, 6.0, 5.8, 6.2]

# Count readings before computing the average.
reading_count = len(thickness_inches)

# Add all readings to get the total.
total_thickness = sum(thickness_inches)

# Divide total by count for average.
average_thickness = total_thickness / reading_count

# Compare each reading with the average.
differences = [value - average_thickness for value in thickness_inches]

# Print a compact engineering summary.
print("Slab thickness readings:", thickness_inches)
print("Number of readings:", reading_count)
print("Total thickness, inches:", round(total_thickness, 2))
print("Average thickness, inches:", round(average_thickness, 2))
print("Differences from average:", [round(value, 2) for value in differences])

```
