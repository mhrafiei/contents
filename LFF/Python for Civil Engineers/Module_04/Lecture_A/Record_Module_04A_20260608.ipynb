## **1. Function Design**

### **1.1. Specific Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1780896307" width="250">



>* Design functions for one clear calculation
>* Make results easier to check and reuse

>* Keep each function focused on one calculation
>* Combine tested functions for larger engineering analyses

>* Make engineering assumptions clear in functions
>* Define limits to prevent calculation misuse



```
#@title Python Code - Specific Calculations

# Functions should answer one engineering question.
# Inputs and outputs make reuse easier.
# Small tests help verify engineering calculations.

# Define a footing area calculation function.
def footing_area(length_m, width_m):
    if length_m <= 0 or width_m <= 0:
        raise ValueError("Footing dimensions must be positive.")

    area_m2 = length_m * width_m
    return area_m2

# Define a slab concrete volume calculation.
def slab_volume(area_m2, thickness_cm):
    if area_m2 <= 0 or thickness_cm <= 0:
        raise ValueError("Area and thickness must be positive.")

    thickness_m = thickness_cm / 100
    volume_m3 = area_m2 * thickness_m
    return volume_m3

# Define axial stress for a member.
def axial_stress(load_kN, area_mm2):
    if area_mm2 <= 0:
        raise ValueError("Area must be positive.")

    load_N = load_kN * 1000
    stress_MPa = load_N / area_mm2
    return stress_MPa

# Run representative engineering test inputs.
area = footing_area(3.0, 2.5)
volume = slab_volume(120.0, 15.0)
stress = axial_stress(250.0, 5000.0)

# Print concise results for reasonableness checks.
print(f"Footing area: {area:.2f} m^2")
print(f"Slab volume: {volume:.2f} m^3")
print(f"Axial stress: {stress:.2f} MPa")

```

### **1.2. Parameter Planning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1780896309" width="250">



>* Choose inputs the calculation truly needs
>* Make assumptions visible and function purpose clear

>* Match inputs to calculation detail
>* Expose choices, avoid unnecessary parameters

>* Reuse functions by changing planned inputs
>* Improve clarity, consistency, testing, and review



```
#@title Python Code - Parameter Planning

# Plan parameters before writing functions.
# Civil calculations need visible assumptions.
# Reusable functions require clear inputs.

# Define a footing volume function.
def footing_volume_cy(length_ft, width_ft, depth_in):
    depth_ft = depth_in / 12
    volume_cf = length_ft * width_ft * depth_ft

    volume_cy = volume_cf / 27
    return volume_cy

# Define a pavement quantity function.
def asphalt_mass_tonnes(area_m2, thickness_cm, density_kg_m3):
    thickness_m = thickness_cm / 100
    volume_m3 = area_m2 * thickness_m

    mass_kg = volume_m3 * density_kg_m3
    return mass_kg / 1000

# Store representative engineering inputs.
footing_cy = footing_volume_cy(12, 8, 18)
asphalt_t = asphalt_mass_tonnes(450, 8, 2400)

# Print concise test results.
print("Footing concrete volume:", round(footing_cy, 2), "cubic yards")
print("Asphalt layer mass:", round(asphalt_t, 2), "tonnes")

# Show parameter planning lessons.
print("Each changing project value became a parameter.")
print("Unit conversions stayed visible inside each function.")

```

### **1.3. Reusable Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1780896310" width="250">



>* Functions reuse common engineering calculation steps
>* Consistent methods improve organization and review

>* Design around the core calculation
>* Let project inputs vary flexibly

>* Balance efficiency with sound engineering judgment
>* Design clear functions for consistent use



```
#@title Python Code - Reusable Calculations

# Reusable functions keep engineering calculations organized.
# Parameters let one method fit many cases.
# Return values support testing and later reuse.

# Define a focused concrete volume function.
def concrete_volume(length_m, width_m, thickness_m):
    if min(length_m, width_m, thickness_m) <= 0:
        raise ValueError("All dimensions must be positive.")

    volume_m3 = length_m * width_m * thickness_m
    return volume_m3

# Define a focused roadway sight distance function.
def stopping_sight_distance(speed_mph, reaction_time_s, deceleration_ft_s2):
    if min(speed_mph, reaction_time_s, deceleration_ft_s2) <= 0:
        raise ValueError("All inputs must be positive.")

    speed_ft_s = speed_mph * 5280 / 3600
    reaction_ft = speed_ft_s * reaction_time_s

    braking_ft = speed_ft_s ** 2 / (2 * deceleration_ft_s2)
    return reaction_ft + braking_ft

# Test the volume function with representative slabs.
sidewalk_m3 = concrete_volume(12.0, 1.5, 0.10)
footing_m3 = concrete_volume(8.0, 2.0, 0.45)

# Test the distance function with roadway inputs.
urban_ft = stopping_sight_distance(35, 2.5, 11.2)
highway_ft = stopping_sight_distance(65, 2.5, 11.2)

# Print concise engineering results for review.
print(f"Sidewalk concrete volume: {sidewalk_m3:.2f} m^3")
print(f"Footing concrete volume: {footing_m3:.2f} m^3")
print(f"Urban stopping sight distance: {urban_ft:.0f} ft")
print(f"Highway stopping sight distance: {highway_ft:.0f} ft")

# Compare outputs against simple expected ranges.
assert 1.7 < sidewalk_m3 < 1.9
assert 7.1 < footing_m3 < 7.3

# Confirm reusable functions returned sensible positives.
assert urban_ft > 0 and highway_ft > urban_ft
print("All representative function tests passed.")

```

## **2. Reusable Engineering Formulas**

### **2.1. Function Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1780896305" width="250">



>* Inputs define needed engineering information
>* Parameter inputs make formulas reusable

>* Choose recognizable, explicit engineering inputs
>* Reuse calculations by changing input values

>* Choose consistent, meaningful input values
>* Design adaptable functions with clear inputs



```
#@title Python Code - Function Inputs

# Functions need clear engineering input values.
# Parameters make formulas reusable across projects.
# Returned values support later engineering decisions.

# Define a reusable footing volume function.
def footing_volume(length_m, width_m, depth_m):
    """Return concrete volume in cubic meters."""
    if min(length_m, width_m, depth_m) <= 0:

        raise ValueError("All footing dimensions must be positive.")
    volume_m3 = length_m * width_m * depth_m
    return volume_m3

# Define a reusable runoff flow function.
def runoff_flow(coefficient, intensity_in_hr, area_acres):
    """Return peak flow using the Rational Method."""
    inputs_valid = 0 <= coefficient <= 1

    if not inputs_valid:
        raise ValueError("Runoff coefficient must be between zero and one.")
    if min(intensity_in_hr, area_acres) <= 0:
        raise ValueError("Rainfall intensity and area must be positive.")

    flow_cfs = coefficient * intensity_in_hr * area_acres
    return flow_cfs

# Call functions using representative engineering inputs.
small_footing = footing_volume(2.4, 1.8, 0.5)
large_footing = footing_volume(3.0, 2.2, 0.6)
parking_runoff = runoff_flow(0.85, 2.1, 4.5)

# Print concise results for comparison.
print(f"Small footing volume: {small_footing:.2f} cubic meters")
print(f"Large footing volume: {large_footing:.2f} cubic meters")
print(f"Parking lot runoff: {parking_runoff:.2f} cubic feet per second")

# Demonstrate same formula with new inputs.
site_runoff = runoff_flow(coefficient=0.55, intensity_in_hr=1.8, area_acres=8.0)
print(f"Mixed site runoff: {site_runoff:.2f} cubic feet per second")

```

### **2.2. Parameters and Returns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1780896302" width="250">



>* Parameters make functions fit many projects
>* Returns provide results for later use

>* Use parameters for changing engineering inputs
>* Return results for later analysis steps

>* Pass only essential engineering inputs
>* Return clear results for decisions



```
#@title Python Code - Parameters and Returns

# Functions make engineering calculations reusable.
# Parameters carry changing project information.
# Returns provide results for later decisions.

# Define a reusable concrete weight function.
def concrete_weight(length_m, width_m, depth_m, unit_weight_kn_m3=24.0):
    """Return concrete element weight in kilonewtons."""

    volume_m3 = length_m * width_m * depth_m
    weight_kn = volume_m3 * unit_weight_kn_m3

    return weight_kn

# Define a reusable runoff calculation function.
def rational_runoff(area_hectares, intensity_mm_hr, runoff_coefficient):
    """Return peak flow in cubic meters per second."""

    flow_m3_s = 0.00278 * runoff_coefficient * intensity_mm_hr * area_hectares
    return flow_m3_s

# Test functions with representative engineering inputs.
beam_weight = concrete_weight(6.0, 0.30, 0.50)
parking_flow = rational_runoff(1.8, 65.0, 0.85)
slab_weight = concrete_weight(4.0, 3.0, 0.15)

# Print concise returned values for review.
print(f"Beam concrete weight: {beam_weight:.2f} kN")
print(f"Parking lot runoff: {parking_flow:.3f} m^3/s")
print(f"Small slab weight: {slab_weight:.2f} kN")
print("Returned values can be stored, compared, or reused.")

```

### **2.3. Returned Calculation Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1780896304" width="250">



>* Return results for later engineering calculations
>* Reusable outputs support checks and reports

>* Return clear, purpose-matched calculation outputs
>* Keep units, precision, and meaning consistent

>* Check returned values against trusted examples
>* Build reusable tools for engineering decisions



```
#@title Python Code - Returned Calculation Results

# Functions return values for later engineering use.
# Returned results support checks and comparisons.
# This example reuses hydraulic calculation outputs.

# Define a reusable runoff flow function.
def runoff_flow(c_factor, intensity_in_per_hr, area_acres):
    flow_cfs = c_factor * intensity_in_per_hr * area_acres
    return flow_cfs

# Define a reusable pipe capacity function.
def pipe_capacity(area_sq_ft, velocity_ft_per_s):
    capacity_cfs = area_sq_ft * velocity_ft_per_s
    return capacity_cfs

# Store representative engineering input values.
site_c = 0.75
rain_i = 2.8
site_area = 4.0

# Store returned runoff result for reuse.
design_flow = runoff_flow(site_c, rain_i, site_area)

# Use the returned result in another check.
pipe_area = 1.8
pipe_velocity = 5.2
pipe_flow = pipe_capacity(pipe_area, pipe_velocity)

# Compare returned values for an engineering decision.
margin = pipe_flow - design_flow
status = "PASS" if margin >= 0 else "REVIEW"

# Print concise results from returned values.
print(f"Returned runoff flow: {design_flow:.2f} cfs")
print(f"Returned pipe capacity: {pipe_flow:.2f} cfs")
print(f"Capacity margin: {margin:.2f} cfs")
print(f"Design check status: {status}")

```

## **3. Testing Functions**

### **3.1. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1780896312" width="250">



>* Choose realistic inputs for engineering calculations
>* Include ordinary, boundary, and error-revealing cases

>* Test routine and edge engineering cases
>* Check expected results with independent reasoning

>* Show valid inputs and expected outputs
>* Build reliable, practical engineering functions



```
#@title Python Code - Representative Test Cases

# Representative tests support reusable engineering functions.
# Each case has an expected hand result.
# Passing tests builds confidence before reuse.

# Define a concrete volume function.
def footing_volume_m3(length_m, width_m, depth_m):
    """Return rectangular footing concrete volume in cubic meters."""

    # Reject impossible physical dimensions.
    if min(length_m, width_m, depth_m) <= 0:
        raise ValueError("All footing dimensions must be positive.")

    # Multiply three metric dimensions.
    volume_m3 = length_m * width_m * depth_m
    return volume_m3

# Store representative engineering test cases.
test_cases = [
    ("small building footing", 2.0, 1.5, 0.4, 1.2),
    ("commercial footing", 4.0, 3.0, 0.8, 9.6),

    ("shallow strip footing", 6.0, 0.6, 0.25, 0.9),
]

# Run tests using a practical tolerance.
passed_count = 0
for name, length_m, width_m, depth_m, expected_m3 in test_cases:

    # Calculate and compare with expected result.
    calculated_m3 = footing_volume_m3(length_m, width_m, depth_m)
    error_m3 = abs(calculated_m3 - expected_m3)

    # Count tests that meet tolerance.
    if error_m3 < 1e-9:
        passed_count += 1

# Test one invalid edge case separately.
try:
    footing_volume_m3(2.0, 1.0, 0.0)
    edge_case_ok = False
except ValueError:
    edge_case_ok = True

# Print concise testing summary.
print("Representative concrete volume tests")
print(f"Routine tests passed: {passed_count} of {len(test_cases)}")
print(f"Invalid zero-depth case handled: {edge_case_ok}")
print("Tests used metric dimensions and hand-checkable results.")

```

### **3.2. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1780896314" width="250">



>* Choose realistic engineering inputs for tests
>* Compare results with trusted independent checks

>* Test typical, boundary, and unusual cases
>* Reject invalid inputs to prevent misleading results

>* Record inputs, outputs, reasons, and interpretations
>* Show tested ranges and prevent misuse



```
#@title Python Code - Representative Test Cases

# Representative tests verify reusable engineering functions.
# Cases should include realistic project conditions.
# Expected values make checking easier.

# Define a reusable footing volume function.
def footing_volume_m3(length_m, width_m, depth_m):
    values = [length_m, width_m, depth_m]

    if min(values) <= 0:
        raise ValueError("All dimensions must be positive.")

    volume_m3 = length_m * width_m * depth_m
    return volume_m3

# Store representative test cases with expectations.
test_cases = [
    ("small pad", 1.5, 1.5, 0.45, 1.0125),
    ("typical footing", 2.4, 1.8, 0.60, 2.5920),
    ("deep footing", 3.0, 2.0, 0.90, 5.4000),
]

# Check each case against independent expected results.
passed_count = 0
for name, length, width, depth, expected in test_cases:
    actual = footing_volume_m3(length, width, depth)

    passed = abs(actual - expected) < 0.0001
    passed_count = passed_count + int(passed)

    status = "PASS" if passed else "CHECK"
    print(f"{name}: {actual:.4f} m^3, expected {expected:.4f}, {status}")

# Test one invalid engineering input safely.
try:
    footing_volume_m3(2.0, 1.5, 0.0)

except ValueError as error:
    print("Invalid case correctly rejected.")

# Summarize how many representative tests passed.
print(f"Representative tests passed: {passed_count}/{len(test_cases)}")

```

### **3.3. Representative Test Cases**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1780896315" width="250">



>* Test realistic engineering situations and limits
>* Choose inputs showing trustworthy calculation behavior

>* Test normal, edge, and invalid inputs
>* Catch unsafe errors before design decisions

>* Explain each test’s purpose and expected result
>* Record assumptions to prove reliable reuse



```
#@title Python Code - Representative Test Cases

# Representative tests build engineering confidence.
# Each case has documented purpose.
# Invalid inputs should fail clearly.

# Define reusable concrete volume calculation.
def concrete_volume_m3(length_m, width_m, thickness_m):
    """Return concrete volume in cubic meters."""

    # Reject dimensions that cannot represent real concrete.
    if min(length_m, width_m, thickness_m) <= 0:
        raise ValueError("All dimensions must be positive meters.")

    # Multiply plan area by thickness.
    return length_m * width_m * thickness_m

# Store representative engineering test cases.
test_cases = [
    {"name": "small slab", "values": (4.0, 3.0, 0.10), "expected": 1.20},

    {"name": "strip footing", "values": (12.0, 0.60, 0.30), "expected": 2.16},

    {"name": "thin repair", "values": (2.0, 1.5, 0.05), "expected": 0.15},
]

# Check calculated answers against documented expectations.
passed_count = 0
for case in test_cases:
    result = concrete_volume_m3(*case["values"])

    # Use tolerance for small rounding differences.
    passed = abs(result - case["expected"]) < 1e-9
    passed_count += int(passed)

    # Print compact evidence for reviewers.
    print(case["name"], "expected", case["expected"], "got", result, "pass", passed)

# Test one invalid case separately.
try:
    concrete_volume_m3(3.0, -2.0, 0.20)

except ValueError as error:
    print("invalid negative width rejected:", str(error))

# Summarize the representative test set.
print("Passed", passed_count, "of", len(test_cases), "valid representative cases.")

```
