## **1. Reading Text Files**

### **1.1. Opening Text Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_01_01.jpg?v=1780897878" width="250">



>* Open files to import engineering data
>* Read mode keeps data unchanged

>* Use correct file names and paths
>* Organize data files consistently for reuse

>* Close files to free system resources
>* Check encoding for reliable text reading



```
#@title Python Code - Opening Text Files

# Create and open engineering text files.
# Use safe file handling patterns.
# Read small rainfall records carefully.

from pathlib import Path

# Store the file path predictably.
data_path = Path("rainfall_station_a.txt")

# Create a tiny engineering data file.
sample_text = "Station,Rainfall_in\nA-101,0.42\nA-101,0.18\n"
data_path.write_text(sample_text, encoding="utf-8")

# Open the file using read mode.
with data_path.open("r", encoding="utf-8") as file_object:
    file_lines = file_object.readlines()

# Count lines read from disk.
record_count = len(file_lines) - 1

# Show the first data line.
first_record = file_lines[1].strip()

# Confirm file reading results.
print("Opened file:", data_path.name)
print("Data records read:", record_count)
print("First rainfall record:", first_record)
print("File automatically closed:", file_object.closed)

```

### **1.2. Line by Line**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_01_02.jpg?v=1780897880" width="250">



>* Read engineering records one line at a time
>* Process large files without loading everything

>* Lines hold text records with varied data
>* Parse and convert values before calculations

>* Spot and handle messy file details
>* Support reliable, traceable engineering data checks



```
#@title Python Code - Line by Line

# Read engineering records line by line.
# Each text line starts as text.
# Small files make examples easy.

# Create a tiny survey text file.
survey_lines = [
    "Station Elevation_m\n",
    "0 102.40\n",
    "25 102.95\n",

    "50 103.10\n",
    "\n",
    "75 103.65\n",
]

# Write sample records to disk.
with open("survey_profile.txt", "w", encoding="utf-8") as file:
    file.writelines(survey_lines)

# Read the file one record.
records = []
with open("survey_profile.txt", "r", encoding="utf-8") as file:
    for line_number, line in enumerate(file, start=1):

        # Remove surrounding spaces and newline.
        clean_line = line.strip()
        if not clean_line:
            continue

        # Skip the descriptive header line.
        if line_number == 1:
            continue

        # Split text fields by spaces.
        parts = clean_line.split()
        if len(parts) != 2:
            continue

        # Convert text values to numbers.
        station_m = float(parts[0])
        elevation_m = float(parts[1])
        records.append((station_m, elevation_m))

# Show concise reading results.
print("Line-by-line reading example")
print(f"Records read: {len(records)}")
print("First record:", records[0])
print("Last record:", records[-1])

# Compute a simple elevation change.
start_elev = records[0][1]
end_elev = records[-1][1]
elevation_change = end_elev - start_elev

# Print one engineering calculation.
print(f"Elevation change: {elevation_change:.2f} m")
print("Each accepted line became one record.")

```

### **1.3. File Reading Basics**

>* Text files start as raw characters
>* Careful reading supports reliable later calculations

>* Open, read, and close files safely
>* Choose full-file or line-by-line reading

>* Check paths, names, modes, and encoding

```
#@title Python Code - File Reading Basics

# Create and inspect a tiny engineering text file.
# Read the file safely using built-in tools.
# Show how raw text becomes usable lines.

# Define a small project file path.
file_path = "concrete_strengths.txt"

# Prepare short text like a lab export.
file_text = "Specimen,Strength_psi\nC1,4100\nC2,4250\nC3,3980\n"

# Write sample data for this demonstration.
with open(file_path, "w", encoding="utf-8") as file:
    file.write(file_text)

# Open the file in reading mode.
with open(file_path, "r", encoding="utf-8") as file:
    all_text = file.read()

# Print a compact view of raw contents.
print("Raw file text preview:")
print(repr(all_text[:60]))

# Read the same file line by line.
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Display basic file reading facts.
print("Number of lines read:", len(lines))
print("Header line:", lines[0].strip())
print("First data line:", lines[1].strip())

# Convert lines into cleaned records.
clean_lines = [line.strip() for line in lines]

# Show cleaned lines without newline characters.
print("Cleaned records:", clean_lines)

```

## **2. Parsing Numeric Data**

### **2.1. Parsing Station Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_02_01.jpg?v=1780897873" width="250">



>* Stations store measurements along engineering alignments
>* Parse text lines into usable numbers

>* Station lines vary by data source
>* Map fields and station IDs consistently

>* Check file data before trusting calculations
>* Validate structure, ranges, and station order



```
#@title Python Code - Parsing Station Data

# Parse station lines from text.
# Convert numeric fields carefully.
# Estimate earthwork volumes simply.

# Store tiny station data.
station_text = """Station CutArea_ft2 FillArea_ft2
0+00 120.0 10.0
1+00 135.0 12.0
2+00 150.0 15.0
3+00 140.0 20.0
"""

# Write text data safely.
file_name = "station_earthwork.txt"
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write(station_text)

# Prepare parsed station records.
records = []
with open(file_name, "r", encoding="utf-8") as file_object:
    header = file_object.readline().strip().split()

    # Read each remaining data line.
    for line_number, line_text in enumerate(file_object, start=2):
        parts = line_text.strip().split()

        # Skip incomplete station rows.
        if len(parts) != 3:
            print(f"Skipped line {line_number}: incomplete row")
            continue

        # Convert station text into feet.
        station_label = parts[0]
        station_feet = int(station_label.split("+")[0]) * 100

        # Convert area fields into floats.
        cut_area = float(parts[1])
        fill_area = float(parts[2])
        records.append((station_label, station_feet, cut_area, fill_area))

# Check enough stations exist.
if len(records) < 2:
    raise ValueError("At least two station rows are required.")

# Estimate volumes between stations.
cut_volume_ft3 = 0.0
fill_volume_ft3 = 0.0
for previous, current in zip(records, records[1:]):
    distance = current[1] - previous[1]

    # Apply average end area method.
    cut_volume_ft3 += distance * (previous[2] + current[2]) / 2
    fill_volume_ft3 += distance * (previous[3] + current[3]) / 2

# Convert cubic feet to cubic yards.
cut_volume_yd3 = cut_volume_ft3 / 27
fill_volume_yd3 = fill_volume_ft3 / 27

# Write concise project results.
output_name = "station_volume_summary.txt"
with open(output_name, "w", encoding="utf-8") as output_file:
    output_file.write(f"Cut volume: {cut_volume_yd3:.1f} yd3\n")
    output_file.write(f"Fill volume: {fill_volume_yd3:.1f} yd3\n")

# Print short visible results.
print(f"Parsed {len(records)} station rows.")
print(f"Columns found: {', '.join(header)}")
print(f"Cut volume: {cut_volume_yd3:.1f} cubic yards")
print(f"Fill volume: {fill_volume_yd3:.1f} cubic yards")
print(f"Results written to {output_name}")

```

### **2.2. Numeric Line Parsing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780897877" width="250">



>* Convert text fields into usable numbers
>* Clean, split, and parse data lines

>* Split each line by its delimiter
>* Convert measurements, keep identifiers as text

>* Handle messy real-world data carefully
>* Flag problems to protect calculation quality



```
#@title Python Code - Numeric Line Parsing

# This script parses numeric station lines.
# Text values become calculation-ready numbers.
# Invalid records are reported safely.

# Store small text lines like file records.
raw_lines = [
    "STA001, 0.0, 24.5, 10.0",
    "STA002, 100.0, 28.0, 12.5",

    "STA003, 200.0, missing, 13.0",
    "STA004, 300.0, 30.5, 15.0",
]

# Prepare containers for parsed records.
valid_records = []
skipped_records = []

# Parse each comma-separated station line.
for line_number, line_text in enumerate(raw_lines, start=1):
    cleaned_line = line_text.strip()

    fields = [part.strip() for part in cleaned_line.split(",")]
    if len(fields) != 4:
        skipped_records.append((line_number, cleaned_line))
        continue

    try:
        station_id = fields[0]
        station_ft = float(fields[1])
        cut_area_ft2 = float(fields[2])
        fill_area_ft2 = float(fields[3])

        valid_records.append((station_id, station_ft, cut_area_ft2, fill_area_ft2))
    except ValueError:
        skipped_records.append((line_number, cleaned_line))

# Calculate average net area from parsed numbers.
net_areas = [record[2] - record[3] for record in valid_records]
average_net_area = sum(net_areas) / len(net_areas)

# Write a short project summary file.
with open("parsed_station_summary.txt", "w") as summary_file:
    summary_file.write("Station parsing summary\n")
    summary_file.write(f"Valid records: {len(valid_records)}\n")
    summary_file.write(f"Skipped records: {len(skipped_records)}\n")

    summary_file.write(f"Average net area: {average_net_area:.2f} ft^2\n")

# Display concise teaching results.
print("Numeric Line Parsing Example")
print(f"Valid station records: {len(valid_records)}")
print(f"Skipped station records: {len(skipped_records)}")
print(f"Average net area: {average_net_area:.2f} ft^2")
print("Summary file created: parsed_station_summary.txt")

```

### **2.3. Handling Header Lines**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_02_03.jpg?v=1780897875" width="250">



>* Headers describe data before measurements begin
>* Separate text lines from numeric calculations

>* Identify header structure before parsing data
>* Skip or test lines intentionally

>* Preserve headers for units and documentation
>* Use column names to verify parsing



```
#@title Python Code - Handling Header Lines

# This example skips descriptive file header lines.
# Numeric station data starts after column labels.
# Parsed values support simple engineering calculations.

# Create a tiny engineering text file.
file_name = "station_areas.txt"

# Store header lines before numeric measurements.
sample_text = """Project: Training Roadway Section
Units: feet and square feet
Station CutArea FillArea
0 12.5 2.0
50 10.0 3.5
100 8.0 5.0
"""

# Write the sample file for this lesson.
with open(file_name, "w", encoding="utf-8") as data_file:
    data_file.write(sample_text)

# Prepare containers for metadata and numbers.
header_lines = []
station_records = []

# Read each line and classify its content.
with open(file_name, "r", encoding="utf-8") as data_file:
    for line in data_file:
        clean_line = line.strip()

        # Skip blank lines safely if present.
        if clean_line == "":
            continue

        # Split each line into separate fields.
        parts = clean_line.split()
        try:
            values = [float(item) for item in parts]
        except ValueError:
            header_lines.append(clean_line)
            continue

        # Store only complete numeric station rows.
        if len(values) == 3:
            station_records.append(values)

# Validate that numeric data was found.
if len(station_records) < 2:
    raise ValueError("Need at least two station rows.")

# Compute average net area between stations.
total_volume = 0.0
for first, second in zip(station_records, station_records[1:]):
    distance = second[0] - first[0]

    # Net area equals fill area minus cut area.
    net_first = first[2] - first[1]
    net_second = second[2] - second[1]
    average_net = (net_first + net_second) / 2

    # Volume uses average end area method.
    total_volume += average_net * distance

# Report concise results for the learner.
print("Header lines detected:", len(header_lines))
print("Numeric station rows:", len(station_records))
print("Estimated net volume:", round(total_volume, 1), "cubic feet")
print("First header line:", header_lines[0])

```

## **3. Writing Project Reports**

### **3.1. Output File Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780897868" width="250">



>* Save calculations as lasting project records
>* Open, write, and close files properly

>* Open files using the correct mode
>* Write or append to protect records

>* Structure files for future readers
>* Include labels, units, and clear results



```
#@title Python Code - Output File Basics

# Create a readable project output file.
# Use write mode for fresh reports.
# Confirm saved content with a preview.

# Store small station-volume results for reporting.
station_results = [("0+00", 125.5), ("1+00", 138.2), ("2+00", 119.8)]

# Choose a clear output filename.
report_file = "station_volume_report.txt"

# Write a formatted plain text report.
with open(report_file, "w", encoding="utf-8") as file:

    # Add a title and units.
    file.write("Station Volume Summary\n")

    file.write("Project: Training Example\n")
    file.write("Units: cubic yards\n\n")

    # Write each station result.
    for station, volume in station_results:

        # Format one readable result line.
        line = f"Station {station}: {volume:.1f} yd^3\n"

        file.write(line)

# Read back only a short preview.
with open(report_file, "r", encoding="utf-8") as file:

    # Keep output safely under fifteen lines.
    preview_lines = file.readlines()[:6]

# Show where the report was saved.
print(f"Created output file: {report_file}")

# Display a concise report preview.
print("Preview of saved project report:")
print("".join(preview_lines))

```

### **3.2. Formatted Output Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780897870" width="250">



>* Make calculation results clear and traceable
>* Use organized reports for review and records

>* Show numbers with appropriate precision and units
>* Align columns for easy review and comparison

>* Design reports for people or programs
>* Use clear structure to prevent misinterpretation



```
#@title Python Code - Formatted Output Files

# Format reports for civil engineering records.
# Write clear columns with practical units.
# Keep output files readable and traceable.

from pathlib import Path

# Store small earthwork results as dictionaries.
records = [
    {"station": "10+00", "cut": 125.4, "fill": 98.2},
    {"station": "10+50", "cut": 88.0, "fill": 132.7},
    {"station": "11+00", "cut": 141.8, "fill": 76.5},
]

# Choose a simple project report filename.
report_path = Path("earthwork_report.txt")

# Compute totals before writing the report.
total_cut = sum(row["cut"] for row in records)
total_fill = sum(row["fill"] for row in records)
net_volume = total_cut - total_fill

# Write a formatted text report safely.
with report_path.open("w", encoding="utf-8") as file:
    file.write("Earthwork Summary Report\n")
    file.write("Project: Training Road Alignment\n\n")
    file.write(f"{'Station':<10}{'Cut yd3':>12}{'Fill yd3':>12}{'Net yd3':>12}\n")

    for row in records:
        net = row["cut"] - row["fill"]
        file.write(f"{row['station']:<10}{row['cut']:>12.1f}{row['fill']:>12.1f}{net:>12.1f}\n")

    file.write("\n")
    file.write(f"Total cut volume:  {total_cut:.1f} cubic yards\n")
    file.write(f"Total fill volume: {total_fill:.1f} cubic yards\n")
    file.write(f"Net project volume: {net_volume:.1f} cubic yards\n")

# Read back only a short preview.
preview_lines = report_path.read_text(encoding="utf-8").splitlines()[:8]
print("Formatted report preview:")
print("\n".join(preview_lines))
print(f"\nSaved report file: {report_path.resolve()}")

```

### **3.3. Saving Report Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780897872" width="250">



>* Save reports as lasting project records
>* Use clear names, folders, and formats

>* Choose create, replace, or append carefully
>* Match saving method to report purpose

>* Use portable plain text report files
>* Include context for clear professional records



```
#@title Python Code - Saving Report Files

# Save reports with clear project information.
# Use text files for durable records.
# Keep outputs small and readable.

from pathlib import Path
from datetime import date

# Store project results in simple records.
project_id = "HWY-24A"
report_date = date.today().isoformat()
output_folder = Path("project_reports")

# Create the folder when needed.
output_folder.mkdir(exist_ok=True)
report_name = f"{project_id}_{report_date}_earthwork_report.txt"
report_path = output_folder / report_name

# Define station volumes in cubic meters.
station_volumes = [(0, 125.4), (50, 138.2), (100, 119.7)]
total_volume = sum(volume for station, volume in station_volumes)
average_volume = total_volume / len(station_volumes)

# Write a formatted plain text report.
with open(report_path, "w", encoding="utf-8") as report_file:
    report_file.write("Earthwork Quantity Report\n")
    report_file.write(f"Project: {project_id}\n")
    report_file.write(f"Created: {report_date}\n")

    report_file.write("Units: cubic meters\n\n")
    report_file.write("Station     Volume\n")
    for station, volume in station_volumes:

        line = f"{station:>7} {volume:>10.1f}\n"
        report_file.write(line)
    report_file.write("\n")
    report_file.write(f"Total volume: {total_volume:.1f}\n")

    report_file.write(f"Average volume: {average_volume:.1f}\n")

# Confirm the saved file location.
print("Report saved successfully.")
print(f"Folder: {output_folder}")
print(f"File: {report_name}")
print(f"Total volume: {total_volume:.1f} cubic meters")

```
