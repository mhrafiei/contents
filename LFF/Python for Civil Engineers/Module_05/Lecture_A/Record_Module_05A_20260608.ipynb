## **1. Calculator Tool Structure**

### **1.1. Input Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_01_01.jpg?v=1780898984" width="250">



>* Inputs define the engineering situation
>* Clear units and labels prevent errors

>* Combine Python concepts for reliable inputs
>* Use clear names and organized data

>* Inputs document assumptions behind each calculation
>* Clear structure supports review, testing, updates



```
#@title Python Code - Input Parameters

# Organize calculator inputs before engineering calculations.
# Use named values with clear units.
# Validate parameters before using formulas.

# Store beam inputs like an engineering data sheet.
beam_inputs = {
    "span_m": 6.0,
    "width_mm": 300.0,
    "depth_mm": 500.0,

    "concrete_mpa": 28.0,
    "steel_mpa": 420.0,
    "service_load_kn_m": 18.0,
}

# Store required units beside each input.
input_units = {
    "span_m": "m",
    "width_mm": "mm",
    "depth_mm": "mm",

    "concrete_mpa": "MPa",
    "steel_mpa": "MPa",
    "service_load_kn_m": "kN/m",
}

# Define simple limits for validation checks.
input_limits = {
    "span_m": (1.0, 30.0),
    "width_mm": (100.0, 1000.0),
    "depth_mm": (150.0, 2000.0),

    "concrete_mpa": (15.0, 70.0),
    "steel_mpa": (250.0, 600.0),
    "service_load_kn_m": (0.1, 200.0),
}

# Check one value against its expected range.
def check_value(name, value, limits):
    lower_limit, upper_limit = limits[name]
    return lower_limit <= value <= upper_limit

# Validate every calculator input before calculations.
validation_results = []
for input_name, input_value in beam_inputs.items():
    value_ok = check_value(input_name, input_value, input_limits)

    validation_results.append((input_name, input_value, value_ok))

# Print a compact input summary for review.
print("Beam calculator input summary")
print("Assumption: simply supported rectangular beam")

# Display each value with units and status.
for input_name, input_value, value_ok in validation_results:
    input_unit = input_units[input_name]
    status_text = "PASS" if value_ok else "CHECK"

    print(f"{input_name}: {input_value} {input_unit} -> {status_text}")

# Confirm whether calculations should safely proceed.
all_inputs_ok = all(result[2] for result in validation_results)
final_status = "READY" if all_inputs_ok else "REVIEW INPUTS"
print(f"Calculator status: {final_status}")

```

### **1.2. Core Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_01_02.jpg?v=1780898982" width="250">



>* Turn inputs into staged engineering results
>* Organize calculations for testing and adaptation

>* Separate reusable logic from project assumptions
>* Name intermediate values with engineering meaning

>* Show calculation steps for easier review
>* Build transparent tools that support engineering judgment



```
#@title Python Code - Core Calculations

# Build a staged beam calculator.
# Validate values before using formulas.
# Report units and pass fail results.

# Store project assumptions as named variables.
span_ft = 18.0
uniform_load_klf = 1.25
allowable_moment_kip_ft = 70.0

# Validate inputs before core calculations.
inputs_are_positive = all([span_ft > 0, uniform_load_klf > 0])
capacity_is_positive = allowable_moment_kip_ft > 0
valid_inputs = inputs_are_positive and capacity_is_positive

# Convert the span for metric context.
span_m = span_ft * 0.3048
load_kn_per_m = uniform_load_klf * 14.5939

# Calculate reactions for a simply supported beam.
reaction_kip = uniform_load_klf * span_ft / 2
max_shear_kip = reaction_kip

# Calculate maximum bending demand at midspan.
max_moment_kip_ft = uniform_load_klf * span_ft ** 2 / 8
usage_ratio = max_moment_kip_ft / allowable_moment_kip_ft

# Check the engineering demand against capacity.
passes_capacity = valid_inputs and usage_ratio <= 1.0
status_text = "PASS" if passes_capacity else "FAIL"

# Prepare short traceable engineering output.
print("Simply supported beam core calculation")
print(f"Span: {span_ft:.1f} ft ({span_m:.2f} m)")
print(f"Uniform load: {uniform_load_klf:.2f} klf ({load_kn_per_m:.2f} kN/m)")
print(f"Reaction each support: {reaction_kip:.2f} kip")
print(f"Maximum shear: {max_shear_kip:.2f} kip")
print(f"Maximum moment: {max_moment_kip_ft:.2f} kip-ft")
print(f"Allowable moment: {allowable_moment_kip_ft:.2f} kip-ft")
print(f"Demand/capacity ratio: {usage_ratio:.2f}")
print(f"Capacity check: {status_text}")

```

### **1.3. Output Workflow**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_01_03.jpg?v=1780898980" width="250">



>* Turn calculations into usable engineering results
>* Show context, units, and interpretation

>* Show conclusions before supporting calculation details
>* Use clear labels, units, and assumptions

>* Flag pass-fail results and unusual conditions.
>* Turn calculations into professional engineering guidance.



```
#@title Python Code - Output Workflow

# Build a concise calculator output workflow.
# Show units, assumptions, and pass fail.
# Keep calculations transparent for reviewers.

# Store known design inputs safely.
footing_load_kn = 820.0
soil_capacity_kpa = 175.0
selected_width_m = 2.3
selected_length_m = 2.4

# Validate important input values.
inputs_positive = all([
    footing_load_kn > 0,
    soil_capacity_kpa > 0,
    selected_width_m > 0,
    selected_length_m > 0,
])

# Calculate required and provided areas.
required_area_m2 = footing_load_kn / soil_capacity_kpa
provided_area_m2 = selected_width_m * selected_length_m
utilization = required_area_m2 / provided_area_m2

# Convert utilization into engineering status.
passes_capacity = utilization <= 1.0
status = "PASS" if passes_capacity else "FAIL"

# Create warnings for review conditions.
warnings = []
if not inputs_positive:
    warnings.append("Check nonpositive input values.")
if utilization > 0.95:
    warnings.append("Capacity margin is less than five percent.")

# Print a compact professional output report.
print("FOOTING BEARING CHECK OUTPUT")
print(f"Status: {status}")
print(f"Required area: {required_area_m2:.2f} m^2")
print(f"Provided area: {provided_area_m2:.2f} m^2")
print(f"Utilization: {utilization:.2%}")
print("Assumption: allowable bearing controls sizing.")

# Summarize warnings without excessive output.
warning_text = "; ".join(warnings) if warnings else "No warnings noted."
print(f"Review notes: {warning_text}")

```

## **2. Input Validation**

### **2.1. Parameter Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_02_01.jpg?v=1780898991" width="250">



>* Check inputs before calculations begin
>* Reject missing, unrealistic, or invalid values

>* Check type, presence, and realistic ranges
>* Confirm inputs fit the engineering system

>* Clear errors guide users to corrections
>* Screened inputs make calculators more reliable



```
#@title Python Code - Parameter Checks

# Parameter checks protect engineering calculator results.
# Small examples reveal common input problems.
# Each check gives a clear message.

# Store sample beam inputs for checking.
beam_inputs = {
    "span_m": 6.0,
    "width_mm": 300,

    "depth_mm": 500,
    "cover_mm": 40,
    "bar_diameter_mm": 20,

    "concrete_mpa": 35,
    "support_type": "simply_supported",
}

# Define accepted support condition names.
allowed_supports = {"simply_supported", "cantilever", "fixed_fixed"}

# Create a list for validation messages.
messages = []

# Check required parameters are present.
required_keys = [
    "span_m", "width_mm", "depth_mm", "cover_mm",

    "bar_diameter_mm", "concrete_mpa", "support_type",
]

# Record missing input names clearly.
missing_keys = [key for key in required_keys if key not in beam_inputs]

# Report missing information before calculations.
if missing_keys:
    messages.append(f"FAIL: missing inputs: {missing_keys}")

# Check numeric parameters are actually numbers.
numeric_keys = [
    "span_m", "width_mm", "depth_mm", "cover_mm",

    "bar_diameter_mm", "concrete_mpa",
]

# Test numeric value type for each field.
for key in numeric_keys:
    value = beam_inputs.get(key, None)

    if not isinstance(value, (int, float)):
        messages.append(f"FAIL: {key} must be numeric")

# Check realistic engineering value ranges.
if 1.0 <= beam_inputs["span_m"] <= 30.0:
    messages.append("PASS: span is within expected building range")

else:
    messages.append("FAIL: span should be 1 to 30 m")

# Check section dimensions are positive.
if beam_inputs["width_mm"] > 0 and beam_inputs["depth_mm"] > 0:
    messages.append("PASS: section dimensions are positive")

else:
    messages.append("FAIL: width and depth must be positive")

# Check concrete strength range.
if 20 <= beam_inputs["concrete_mpa"] <= 60:
    messages.append("PASS: concrete strength is realistic")

else:
    messages.append("FAIL: concrete strength should be 20 to 60 MPa")

# Compute effective depth only after basic checks.
effective_depth = beam_inputs["depth_mm"] - beam_inputs["cover_mm"]

# Check relationship between cover and depth.
if effective_depth > beam_inputs["bar_diameter_mm"]:
    messages.append("PASS: effective depth leaves room for reinforcement")

else:
    messages.append("FAIL: cover leaves insufficient reinforcement space")

# Check support category against accepted choices.
if beam_inputs["support_type"] in allowed_supports:
    messages.append("PASS: support type is recognized")

else:
    messages.append("FAIL: support type is not recognized")

# Print a compact validation report.
print("Input Validation Report")

# Print each validation message.
for message in messages:
    print(message)

# Print the intermediate engineering value.
print(f"Effective depth = {effective_depth} mm")

```

### **2.2. Value Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_02_02.jpg?v=1780898994" width="250">



>* Check numbers are reasonable, not just valid
>* Catch unrealistic inputs before wrong results

>* Check positive values and practical ranges
>* Flag entries that violate engineering sense

>* Check intermediate results before errors spread
>* Flag suspicious values for engineering review



```
#@title Python Code - Value Checks

# Value checks protect engineering calculations.
# This example checks beam inputs.
# Intermediate results are checked too.

# Store sample beam inputs with units.
span_m = 6.0
load_kN_m = 18.0
allowable_MPa = 12.0
width_mm = 300.0
height_mm = 500.0

# Collect validation messages before calculating.
issues = []
if not 0.5 <= span_m <= 30.0:
    issues.append("Span outside practical meter range")

# Check load and material assumptions.
if not 0.1 <= load_kN_m <= 200.0:
    issues.append("Load outside practical kN/m range")
if not 1.0 <= allowable_MPa <= 40.0:
    issues.append("Allowable stress outside MPa range")

# Check section dimensions for reasonableness.
if not 50.0 <= width_mm <= 2000.0:
    issues.append("Width outside practical millimeter range")
if not 50.0 <= height_mm <= 3000.0:
    issues.append("Height outside practical millimeter range")

# Convert section dimensions into meters.
width_m = width_mm / 1000.0
height_m = height_mm / 1000.0
section_modulus_m3 = width_m * height_m**2 / 6.0

# Calculate simple-span maximum moment.
moment_kNm = load_kN_m * span_m**2 / 8.0
stress_MPa = moment_kNm / section_modulus_m3 / 1000.0

# Check key intermediate results.
if section_modulus_m3 <= 0.0:
    issues.append("Section modulus must be positive")
if stress_MPa <= 0.0 or stress_MPa > 100.0:
    issues.append("Calculated stress looks suspicious")

# Select pass-fail status safely.
passes_value_checks = len(issues) == 0
passes_design = passes_value_checks and stress_MPa <= allowable_MPa

# Print a compact engineering report.
print("Beam value-check report")
print(f"Span = {span_m:.2f} m, load = {load_kN_m:.1f} kN/m")
print(f"Section = {width_mm:.0f} mm x {height_mm:.0f} mm")
print(f"Maximum moment = {moment_kNm:.1f} kN*m")
print(f"Calculated stress = {stress_MPa:.2f} MPa")

# Print assumptions and validation status.
print(f"Allowable stress = {allowable_MPa:.2f} MPa")
print("Assumption: simply supported beam, uniform load")
print(f"Value checks passed: {passes_value_checks}")
print(f"Design pass: {passes_design}")
print("Issues: " + ("none" if not issues else "; ".join(issues)))

```

### **2.3. Moment Validation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_02_03.jpg?v=1780898993" width="250">



>* Check moments where calculations create or use them
>* Confirm results match realistic beam behavior

>* Check moments against expected beam behavior
>* Keep sign conventions consistent for design

>* Flag unrealistic moments with helpful diagnostic warnings
>* Keep engineering judgment visible in automation



```
#@title Python Code - Moment Validation

# Validate moments before final design checks.
# Use simple beam formulas carefully.
# Report units and engineering warnings.

import math

# Store beam inputs in consistent units.
span_m = 6.0
load_kN = 24.0
load_position_m = 3.0

# Define a small validation helper.
def add_check(checks, passed, message):
    status = "PASS" if passed else "FAIL"
    checks.append(f"{status}: {message}")

# Collect validation messages for review.
checks = []
add_check(checks, span_m > 0, "span is positive")
add_check(checks, load_kN > 0, "gravity load is positive")
add_check(checks, 0 <= load_position_m <= span_m, "load lies on beam")

# Stop only if geometry is impossible.
valid_geometry = span_m > 0 and 0 <= load_position_m <= span_m
if valid_geometry:
    reaction_left_kN = load_kN * (span_m - load_position_m) / span_m
    reaction_right_kN = load_kN * load_position_m / span_m
else:
    reaction_left_kN = math.nan
    reaction_right_kN = math.nan

# Compute maximum moment for point load.
if valid_geometry:
    max_moment_kNm = reaction_left_kN * load_position_m
else:
    max_moment_kNm = math.nan

# Validate intermediate engineering behavior.
add_check(checks, reaction_left_kN >= 0, "left reaction is not uplift")
add_check(checks, reaction_right_kN >= 0, "right reaction is not uplift")
add_check(checks, abs(reaction_left_kN + reaction_right_kN - load_kN) < 1e-9, "vertical equilibrium closes")
add_check(checks, max_moment_kNm >= 0, "sagging moment sign is expected")

# Compare against a simple upper bound.
reference_moment_kNm = load_kN * span_m / 4
reasonable_limit_kNm = 1.10 * reference_moment_kNm
add_check(checks, max_moment_kNm <= reasonable_limit_kNm, "moment is within expected range")

# Print concise calculator output.
print("Moment validation example: simply supported beam")
print(f"Span = {span_m:.2f} m, Load = {load_kN:.2f} kN")
print(f"Load position = {load_position_m:.2f} m from left support")
print(f"Reactions = {reaction_left_kN:.2f}, {reaction_right_kN:.2f} kN")
print(f"Maximum moment = {max_moment_kNm:.2f} kN-m")

# Print each validation result.
for item in checks:
    print(item)

# Create a final pass indicator.
overall_pass = all(item.startswith("PASS") for item in checks)
print("Overall result:", "PASS" if overall_pass else "REVIEW REQUIRED")

```

## **3. Result Reporting**

### **3.1. Units and Assumptions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_03_01.jpg?v=1780898986" width="250">



>* Always show values with clear units
>* Make units visible wherever results are used

>* Choose units that match engineering practice
>* Round results to useful, honest precision

>* State assumptions alongside final calculated results
>* Show when results are valid and limited



```
#@title Python Code - Units and Assumptions

# Report engineering results with clear units.
# State assumptions beside calculated design values.
# Use pass fail wording for decisions.

# Define calculator inputs in metric units.
load_kn = 650.0
footing_width_m = 1.80
footing_length_m = 2.20
allowable_bearing_kpa = 180.0

# Store assumptions near reported results.
assumptions = [
    "uniform soil pressure",
    "concentric vertical loading",
    "service load combination",
    "allowable bearing pressure basis",
]

# Validate positive engineering inputs.
values_are_positive = all(
    value > 0 for value in [
        load_kn,
        footing_width_m,
        footing_length_m,
        allowable_bearing_kpa,
    ]
)

# Stop calculation if validation fails.
if not values_are_positive:
    raise ValueError("All input values must be positive.")

# Calculate plan area and soil pressure.
footing_area_m2 = footing_width_m * footing_length_m
soil_pressure_kpa = load_kn / footing_area_m2
utilization_ratio = soil_pressure_kpa / allowable_bearing_kpa

# Create a readable pass fail indicator.
status_text = "PASS" if soil_pressure_kpa <= allowable_bearing_kpa else "FAIL"

# Print a compact engineering result report.
print("Footing bearing check report")
print(f"Applied service load: {load_kn:.1f} kN")
print(f"Footing plan area: {footing_area_m2:.2f} m^2")
print(f"Calculated soil pressure: {soil_pressure_kpa:.1f} kPa")
print(f"Allowable soil pressure: {allowable_bearing_kpa:.1f} kPa")
print(f"Utilization ratio: {utilization_ratio:.2f}")
print(f"Decision: {status_text}")
print("Assumptions: " + "; ".join(assumptions))

```

### **3.2. Assumptions Summary**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_03_02.jpg?v=1780898987" width="250">



>* Assumptions explain when results are valid
>* Clear reporting supports review and revisions

>* State specific assumptions in plain language
>* Separate inputs, defaults, and simplifications

>* Visible assumptions support review and accountability
>* Clear context makes results defensible and verifiable



```
#@title Python Code - Assumptions Summary

# This script reports calculator assumptions clearly.
# It links assumptions to engineering results.
# Pass fail wording supports quick review.

# Define compact project inputs.
span_m = 6.0
service_load_kn_m = 18.0
allowable_moment_knm = 95.0

# State modeling assumptions explicitly.
assumptions = [
    "Beam treated as simply supported",
    "Uniform service load applied along span",
    "Self weight included in service load",
    "Allowable moment based on selected design check",
]

# Validate important calculation inputs.
inputs_valid = span_m > 0 and service_load_kn_m >= 0
capacity_valid = allowable_moment_knm > 0

# Compute maximum moment for uniform loading.
max_moment_knm = service_load_kn_m * span_m**2 / 8
utilization = max_moment_knm / allowable_moment_knm

# Build a clear pass fail indicator.
status = "PASS" if inputs_valid and capacity_valid and utilization <= 1 else "FAIL"

# Print concise engineering output.
print("RESULT REPORT: Simply Supported Beam Check")
print(f"Maximum moment: {max_moment_knm:.1f} kN·m")
print(f"Allowable moment: {allowable_moment_knm:.1f} kN·m")
print(f"Utilization ratio: {utilization:.2f}")
print(f"Status: {status}")

# Print assumptions near final results.
print("Assumptions summary:")
for number, text in enumerate(assumptions, start=1):
    print(f"{number}. {text}")

```

### **3.3. Pass Fail Indicators**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_A/image_03_03.jpg?v=1780898989" width="250">



>* Show whether results meet design limits
>* Help users interpret calculations quickly

>* Report values, limits, units, and status
>* Show which checks control redesign

>* Show passes depend on stated assumptions
>* Use indicators as review support only



```
#@title Python Code - Pass Fail Indicators

# Pass fail indicators support engineering decisions.
# Each check reports values and units.
# Assumptions keep the result traceable.

# Store stated design assumptions clearly.
pipe_id = "Culvert A"
required_flow_cfs = 42.0
provided_flow_cfs = 45.5

# Define a margin for near limit warnings.
review_margin_percent = 5.0
ratio = provided_flow_cfs / required_flow_cfs
margin_percent = (ratio - 1.0) * 100.0

# Validate intermediate results before reporting.
values_are_positive = required_flow_cfs > 0 and provided_flow_cfs > 0
result_is_near_limit = 0.0 <= margin_percent < review_margin_percent

# Choose a clear engineering status label.
if not values_are_positive:
    status_text = "INVALID INPUTS"
elif provided_flow_cfs >= required_flow_cfs and result_is_near_limit:
    status_text = "PASS, REVIEW"
elif provided_flow_cfs >= required_flow_cfs:
    status_text = "PASS"
else:
    status_text = "FAIL"

# Present the calculation as an engineering report.
print(f"Check: {pipe_id} stormwater capacity")
print(f"Required flow = {required_flow_cfs:.1f} cfs")
print(f"Provided flow = {provided_flow_cfs:.1f} cfs")
print(f"Capacity ratio = {ratio:.2f}")
print(f"Margin = {margin_percent:.1f}%")
print(f"Status = {status_text}")
print("Assumption: steady peak flow only, no erosion check.")

```
