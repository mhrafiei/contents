```
#@title Python Code - Problem Definition

# Define a focused civil workflow.
# Keep inputs and outputs clear.
# Score ideas before coding begins.

# Store candidate automation project ideas.
project_ideas = [
    {
        "name": "Pipe slope checker",
        "input": "pipe lengths and invert elevations",
        "output": "minimum slope review table",
        "pain": "reduces manual arithmetic errors",
        "scope": "narrow",
    },
    {
        "name": "Automate structural design",
        "input": "all building design information",
        "output": "complete sealed design package",
        "pain": "too broad for student script",
        "scope": "broad",
    },
    {
        "name": "Traffic count summarizer",
        "input": "hourly vehicle counts",
        "output": "peak hour and daily total",
        "pain": "standardizes repeated summaries",
        "scope": "narrow",
    },
]

# Define simple planning criteria.
criteria = ["clear_input", "clear_output", "real_pain", "narrow_scope"]

# Score one idea against planning criteria.
def score_project(idea):
    score = 0
    score += 1 if idea["input"] else 0
    score += 1 if idea["output"] else 0
    score += 1 if "reduces" in idea["pain"] or "standardizes" in idea["pain"] else 0
    score += 1 if idea["scope"] == "narrow" else 0
    return score

# Evaluate each proposed capstone idea.
results = []
for idea in project_ideas:
    score = score_project(idea)
    decision = "good candidate" if score >= 3 else "revise definition"
    results.append((idea["name"], score, decision))

# Print a compact planning review.
print("Problem definition screening:")
for name, score, decision in results:
    print(f"- {name}: {score}/4, {decision}")

# Select the strongest beginner project.
best_project = max(results, key=lambda item: item[1])
print(f"Recommended focus: {best_project[0]}")

```

```
#@title Python Code - Workflow Requirements

# Plan requirements before coding automation.
# Civil workflows need clear boundaries.
# Small examples can reveal scope.

# Store mock inspection records as dictionaries.
records = [
    {"asset": "P-101", "date": "2026-04-01", "defect": "crack", "severity": 3},
    {"asset": "P-102", "date": "2026-04-01", "defect": "sediment", "severity": 2},
    {"asset": "P-101", "date": "", "defect": "joint gap", "severity": 4},
    {"asset": "P-103", "date": "2026-04-02", "defect": "", "severity": 1},
]

# Define required fields for every record.
required_fields = ["asset", "date", "defect", "severity"]

# Define workflow requirements in plain language.
requirements = [
    "Accept inspection records with asset, date, defect, and severity.",
    "Flag blank required fields before summarizing results.",
    "Group severe defects by asset for engineer review.",
]

# Check records for blank required values.
issues = []
for position, record in enumerate(records, start=1):
    for field in required_fields:
        value = record.get(field, "")
        if value == "":
            issues.append((position, field))

# Summarize severe defects by asset identifier.
follow_up_assets = {}
for record in records:
    severity = record.get("severity", 0)
    asset = record.get("asset", "UNKNOWN")
    if severity >= 3:
        follow_up_assets[asset] = follow_up_assets.get(asset, 0) + 1

# Print a compact project planning checklist.
print("Workflow requirement check for storm drain inspections")
print("Records reviewed:", len(records))
print("Requirements listed:", len(requirements))
print("Blank required fields found:", len(issues))
print("Assets needing follow-up:", sorted(follow_up_assets))

# Show limited details for quick verification.
if issues:
    print("First data issue:", issues[0])
else:
    print("First data issue: none")

# Display one requirement that guided the script.
print("Example requirement:", requirements[1])

```

```
#@title Python Code - Engineering Scope

# This example scopes a small engineering automation project.
# It separates inputs, assumptions, outputs, and exclusions.
# It uses only built-in Python data structures.

# Define the proposed civil engineering workflow.
project_name = "Concrete volume planning helper"
workflow_area = "Preliminary quantity takeoff"
engineering_purpose = "Estimate simple rectangular concrete volumes"

# Store practical project inputs.
input_fields = ["element", "length_ft", "width_ft", "depth_in"]
required_users = ["engineer", "estimator", "project manager"]
review_stage = "preliminary planning only"

# Define small sample elements.
elements = [
    {"element": "Sidewalk panel", "length_ft": 12.0, "width_ft": 5.0, "depth_in": 4.0},
    {"element": "Equipment pad", "length_ft": 8.0, "width_ft": 6.0, "depth_in": 8.0},
]

# Validate the sample dataset size.
if len(elements) == 0 or len(elements) > 10:
    raise ValueError("Element list must contain one to ten items.")

# Define transparent engineering assumptions.
assumptions = [
    "Elements are rectangular prisms.",
    "Dimensions are field-entered values.",
    "Waste, reinforcement, and formwork are excluded.",
]

# Calculate volume for each element.
for item in elements:
    depth_ft = item["depth_in"] / 12.0
    item["volume_cuft"] = item["length_ft"] * item["width_ft"] * depth_ft

# Summarize useful engineering outputs.
total_cuft = sum(item["volume_cuft"] for item in elements)
total_cuyd = total_cuft / 27.0
largest = max(elements, key=lambda item: item["volume_cuft"])

# Print concise scope and result information.
print("Project:", project_name)
print("Workflow:", workflow_area)
print("Purpose:", engineering_purpose)
print("Inputs:", ", ".join(input_fields))

# Print assumptions without excessive output.
print("Assumptions:", " | ".join(assumptions))
print("Review stage:", review_stage)
print("Elements checked:", len(elements))
print("Total volume, cubic yards:", round(total_cuyd, 2))

# Print the primary review flag.
print("Largest item:", largest["element"])
print("Use result as aid, not final certification.")

```

### **2.1. Core Script Structure**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_B/image_02_01.jpg?v=1780900698" width="250">



>* Organize scripts from inputs to outputs
>* Make engineering logic clear and traceable

>* Separate inputs, validation, calculations, and outputs
>* Organized sections improve testing and accountability

>* Start clearly and follow a predictable workflow
>* Use built-ins to make reasoning traceable



```
#@title Python Code - Core Script Structure

# Structure supports clear engineering automation.
# Built-ins can organize practical calculations.
# Functions separate inputs checks and reporting.

# Define small drainage project records.
locations = [
    {"name": "Inlet A", "area_ac": 3.2, "c": 0.72, "intensity_in_hr": 4.1},
    {"name": "Inlet B", "area_ac": 2.4, "c": 0.65, "intensity_in_hr": 3.8},
    {"name": "Inlet C", "area_ac": 4.0, "c": 0.80, "intensity_in_hr": 4.4},
]

# Validate required fields and ranges.
def validate_location(record, required_fields):
    missing = [field for field in required_fields if field not in record]
    values_are_positive = all(record.get(field, 0) > 0 for field in required_fields[1:])
    coefficient_ok = 0 <= record.get("c", -1) <= 1
    return (not missing) and values_are_positive and coefficient_ok

# Compute rational method design flow.
def compute_flow_cfs(record):
    area_acres = record["area_ac"]
    coefficient = record["c"]
    intensity = record["intensity_in_hr"]
    return coefficient * intensity * area_acres

# Build report rows from valid inputs.
def build_flow_report(records):
    required = ["name", "area_ac", "c", "intensity_in_hr"]
    rows = []
    for record in records:
        if validate_location(record, required):
            flow_value = compute_flow_cfs(record)
            rows.append((record["name"], round(flow_value, 2)))
    return rows

# Run the workflow through one entry point.
def main_workflow():
    report_rows = build_flow_report(locations)
    print("Drainage design flow summary")
    for name, flow in report_rows:
        print(f"{name}: {flow} cfs")
    print(f"Valid locations processed: {len(report_rows)}")
    return report_rows

# Execute the organized capstone script.
final_report_rows = main_workflow()

```

```
#@title Python Code - Built In Data Structures

# Build a small drainage review model.
# Use only built in structures.
# Summarize checks for engineering decisions.

# Store fixed design criteria in dictionaries.
design_limits = {"min_cover_m": 1.2, "max_runoff": 0.95}

# Store standard pipe options as tuples.
pipe_options = ((300, "PVC"), (450, "RCP"), (600, "RCP"))

# Store catchments as labeled dictionaries.
catchments = [
    {"id": "C1", "area_ha": 1.8, "c": 0.55, "pipe_mm": 300},
    {"id": "C2", "area_ha": 3.2, "c": 0.72, "pipe_mm": 450},
]

# Add a catchment with possible review concern.
catchments.append({"id": "C3", "area_ha": 2.5, "c": 0.98, "pipe_mm": 300})

# Create useful lookups from built in records.
allowed_sizes = {size for size, material in pipe_options}

# Track unique pipe materials for reporting.
unique_materials = {material for size, material in pipe_options}

# Initialize summary values for review.
total_area = 0.0
flagged_ids = []

# Review each catchment using readable keys.
for item in catchments:
    total_area += item["area_ha"]
    if item["c"] > design_limits["max_runoff"]:
        flagged_ids.append(item["id"])
    if item["pipe_mm"] not in allowed_sizes:
        flagged_ids.append(item["id"])

# Prepare compact report values.
project_record = ("Drainage capstone", len(catchments), round(total_area, 2))

# Print a concise engineering summary.
print("Project:", project_record[0])
print("Catchments reviewed:", project_record[1])
print("Total area, hectares:", project_record[2])
print("Allowed pipe sizes, mm:", sorted(allowed_sizes))
print("Pipe materials found:", sorted(unique_materials))
print("Catchments needing review:", flagged_ids or "None")

```

```
#@title Python Code - Built In Data Structures

# Built in structures organize engineering project data.
# This example checks a culvert inspection workflow.
# Lists, tuples, dictionaries, and sets are used.

# Store fixed project reference information.
project_location = ("Road A", "Station 10+50")

# Store culvert records in a review order.
culverts = [
    {"id": "C1", "length_ft": 42, "diameter_in": 36},
    {"id": "C2", "length_ft": 55, "diameter_in": 48},
    {"id": "C3", "length_ft": 38, "diameter_in": 30},
]

# Store condition ratings by asset identifier.
condition_by_id = {"C1": "good", "C2": "needs cleaning"}

# Store submitted inspection form identifiers.
submitted_forms = {"C1", "C2"}

# Find all culvert identifiers from records.
expected_forms = {record["id"] for record in culverts}

# Compare sets to identify missing forms.
missing_forms = sorted(expected_forms - submitted_forms)

# Compute total pipe length safely.
total_length_ft = sum(record["length_ft"] for record in culverts)

# Build simple status records for reporting.
status_lines = []
for record in culverts:
    asset_id = record["id"]
    condition = condition_by_id.get(asset_id, "not rated")
    status_lines.append((asset_id, condition))

# Print a compact engineering summary.
print("Project:", project_location[0], project_location[1])
print("Culverts reviewed:", len(culverts))
print("Total pipe length, ft:", total_length_ft)
print("Missing inspection forms:", missing_forms if missing_forms else "none")
print("Condition status records:", status_lines)

```

### **3.1. Technical Validation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_B/image_03_01.jpg?v=1780900703" width="250">



>* Confirm results match the engineering task

```
#@title Python Code - Technical Validation

# Technical validation checks engineering automation results.
# Built-in Python supports transparent review steps.
# Small test cases reveal script weaknesses.

# No extra installations are required.

# Define footing test cases with expected volumes.
test_cases = [
    {"name": "ordinary", "length_ft": 10.0, "width_ft": 8.0, "depth_ft": 1.5, "expected_ft3": 120.0},
    {"name": "zero depth", "length_ft": 10.0, "width_ft": 8.0, "depth_ft": 0.0, "expected_ft3": 0.0},
]

# Add one problematic case for safe failure.
bad_case = {"name": "negative depth", "length_ft": 10.0, "width_ft": 8.0, "depth_ft": -1.0}
tolerance = 0.001
validated_results = []

# Create a small reusable calculation function.
def footing_volume(case):
    values = [case["length_ft"], case["width_ft"], case["depth_ft"]]

    # Reject dimensions that cannot represent a footing.
    if any(value < 0 for value in values):
        raise ValueError("Footing dimensions must be nonnegative.")

    # Return cubic feet from consistent foot dimensions.
    return case["length_ft"] * case["width_ft"] * case["depth_ft"]

# Check ordinary and edge cases against trusted answers.
for case in test_cases:
    calculated = footing_volume(case)

    # Compare calculated output against hand calculation.
    passed = abs(calculated - case["expected_ft3"]) <= tolerance
    validated_results.append((case["name"], calculated, passed))

# Confirm problematic input fails safely and clearly.
try:
    footing_volume(bad_case)

    # This line records an unsafe validation failure.
    bad_case_message = "negative depth: failed unsafe"
except ValueError:
    bad_case_message = "negative depth: passed safe failure"

# Summarize validation without excessive printed output.
print("Technical validation summary")
print("Checked cases:", len(test_cases) + 1)

# Print each trusted comparison result compactly.
for name, volume, passed in validated_results:
    status = "PASS" if passed else "FAIL"
    print(name, round(volume, 3), "ft^3", status)

# Print the safe failure result last.
print(bad_case_message)

```

### **3.2. Code Clarity**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_B/image_03_02.jpg?v=1780900707" width="250">



>* Make script purpose and workflow easy to follow

>* Explain assumptions, not obvious code steps
>* Keep comments concise, current, and helpful

>* Clear structure reduces errors and eases updates

```
#@title Python Code - Code Clarity

# Review code clarity using a drainage slope example.
# Clear names help engineers trust automated checks.
# This script stays small, readable, and defensive.

# Store pipe survey records as simple dictionaries.
pipe_records = [
    {"pipe_id": "P-101", "length_m": 45.0, "drop_m": 0.42},
    {"pipe_id": "P-102", "length_m": 28.0, "drop_m": 0.10},
    {"pipe_id": "P-103", "length_m": 36.0, "drop_m": 0.55},
]

# Define the engineering limit near the calculation.
minimum_slope_percent = 1.0

# Validate input size before calculating results.
record_count = len(pipe_records)
if record_count == 0:
    raise ValueError("No pipe records were provided.")

# Create a small helper for readable slope calculations.
def calculate_slope_percent(length_m, drop_m):
    if length_m <= 0:
        raise ValueError("Pipe length must be positive.")

    return (drop_m / length_m) * 100

# Create another helper for readable pass fail labels.
def classify_slope(slope_percent, required_percent):
    if slope_percent >= required_percent:
        return "OK"

    return "CHECK"

# Calculate reviewed results using clear output names.
reviewed_results = []
for pipe in pipe_records:
    slope_percent = calculate_slope_percent(pipe["length_m"], pipe["drop_m"])
    status = classify_slope(slope_percent, minimum_slope_percent)

    reviewed_results.append((pipe["pipe_id"], slope_percent, status))

# Print compact results for quick engineering review.
print("Drainage pipe slope review")
print(f"Minimum required slope: {minimum_slope_percent:.1f}%")
print("Pipe    Slope (%)   Status")

# Keep output short and aligned for readability.
for pipe_id, slope_percent, status in reviewed_results:
    print(f"{pipe_id:<7} {slope_percent:>7.2f}   {status}")

# Summarize whether the script output is actionable.
check_count = sum(1 for item in reviewed_results if item[2] == "CHECK")
print(f"Pipes needing review: {check_count} of {record_count}")

```

### **3.3. Engineering Usefulness**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_05/Lecture_B/image_03_03.jpg?v=1780900705" width="250">



>* Reduce real engineering work and errors

>* Match inputs, logic, and outputs to workflows

>* Works beyond one test case

```
#@title Python Code - Engineering Usefulness

# Review usefulness for civil engineering automation.
# Connect outputs to practical decisions.
# Keep data tiny and readable.

# Store sample culvert inspection records.
records = [
    {"id": "C-101", "rating": 4, "length_m": 18.0, "debris": "yes"},
    {"id": "C-102", "rating": 2, "length_m": 12.5, "debris": "no"},

    {"id": "C-103", "rating": None, "length_m": 20.0, "debris": "yes"},
    {"id": "C-104", "rating": 1, "length_m": 9.5, "debris": "no"},
]

# Define practical review thresholds.
minimum_rating = 3
priority_debris = "yes"
missing_count = 0

# Prepare decision focused outputs.
priority_ids = []
checked_count = 0
total_length = 0.0

# Review each inspection record carefully.
for record in records:
    rating = record.get("rating")
    length_m = record.get("length_m", 0.0)

    # Track missing values before decisions.
    if rating is None:
        missing_count += 1
        continue

    checked_count += 1
    total_length += length_m

    # Flag locations needing engineering attention.
    if rating < minimum_rating or record.get("debris") == priority_debris:
        priority_ids.append(record.get("id", "unknown"))

# Calculate a simple workflow summary.
average_length = total_length / checked_count if checked_count else 0.0
useful_when = bool(priority_ids) and missing_count >= 0

# Print concise engineering usefulness review.
print("Engineering usefulness review")
print(f"Records checked: {checked_count} usable, {missing_count} incomplete")
print(f"Priority culverts: {', '.join(priority_ids)}")
print(f"Average checked length: {average_length:.1f} m")
print(f"Useful decision support: {useful_when}")

```
