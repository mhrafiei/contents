## **1. For Loops**

### **1.1. Sequence Iteration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780893009" width="250">



>* For loops process values one by one
>* Repeat engineering calculations consistently across sequences

>* Process meaningful engineering value sequences
>* Focus on calculation logic, not counting

>* Process every engineering value consistently
>* Plan data, repeated actions, and results



```
#@title Python Code - Sequence Iteration

# Demonstrate sequence iteration with engineering values.
# Each loop step processes one rainfall intensity.
# Results stay short for beginner readability.

# Store rainfall intensities in millimeters per hour.
rainfall_mm_hr = [20, 35, 50, 65]

# Store watershed area in square kilometers.
area_km2 = 1.2

# Store runoff coefficient for paved urban ground.
runoff_coefficient = 0.75

# Introduce the repeated calculation.
print("Runoff estimates from sequence iteration:")

# Iterate through each engineering input value.
for intensity in rainfall_mm_hr:
    # Compute Rational Method flow in cubic meters per second.
    flow_m3_s = 0.278 * runoff_coefficient * intensity * area_km2

    # Report the current input and calculated result.
    print(f"Rainfall {intensity:>2} mm/hr -> Flow {flow_m3_s:5.2f} m^3/s")

# Confirm the loop handled every sequence value.
processed_count = len(rainfall_mm_hr)

# Print a compact summary line.
print(f"Processed {processed_count} rainfall values using one for loop.")

```

### **1.2. Iterating Engineering Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780893011" width="250">



>* Repeat calculations across engineering value sequences
>* Reduce rewriting and improve consistency

>* Apply one method across changing engineering data
>* Treat each value consistently for reliable results

>* Repeat reviews across many design cases
>* Keep engineering logic consistent as data grows



```
#@title Python Code - Iterating Engineering Values

# For loops repeat engineering calculations.
# Each value receives identical treatment.
# Results stay organized and comparable.

# Define beam span values in meters.
span_meters = [4.0, 5.5, 7.0, 8.5]

# Store calculated deflection estimates here.
deflection_estimates = []

# Loop through each engineering span value.
for span in span_meters:
    # Estimate deflection using a simple teaching formula.
    deflection_mm = 0.08 * span ** 2

    # Save each result for later review.
    deflection_estimates.append(deflection_mm)

    # Print one concise result line.
    print(f"Span {span:.1f} m gives {deflection_mm:.2f} mm.")

# Count how many design cases were checked.
case_count = len(deflection_estimates)

# Find the largest calculated deflection.
maximum_deflection = max(deflection_estimates)

# Print a short summary for engineers.
print(f"Checked {case_count} spans using one loop.")
print(f"Maximum estimated deflection is {maximum_deflection:.2f} mm.")

```

### **1.3. Looping Engineering Sequences**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780893013" width="250">



>* Apply calculations across engineering data sequences
>* Reduce repetition, errors, and update effort

>* Repeat similar calculations across engineering sites
>* Apply judgment consistently to value sequences

>* Know what each sequence item represents
>* Keep results traceable to engineering inputs



```
#@title Python Code - Looping Engineering Sequences

# For loops repeat engineering calculations.
# Sequences hold related project values.
# Each item gets processed consistently.

# Store subcatchment areas in acres.
areas_acres = [1.2, 2.5, 0.9, 3.1]

# Store runoff coefficients for each area.
runoff_coefficients = [0.45, 0.60, 0.35, 0.70]

# Use one design rainfall intensity.
rainfall_in_per_hour = 2.1

# Prepare storage for computed runoff values.
runoff_results_cfs = []

# Loop through paired engineering sequences.
for area, coefficient in zip(areas_acres, runoff_coefficients):
    # Apply the Rational Method formula.
    runoff_cfs = coefficient * rainfall_in_per_hour * area

    # Save each computed runoff result.
    runoff_results_cfs.append(runoff_cfs)

# Count subcatchments checked by the loop.
subcatchment_count = len(runoff_results_cfs)

# Accumulate total runoff from all areas.
total_runoff_cfs = sum(runoff_results_cfs)

# Find the largest individual runoff value.
maximum_runoff_cfs = max(runoff_results_cfs)

# Print compact traceable engineering results.
print("Runoff values computed for each subcatchment:")

# Show each loop result with numbering.
for index, runoff_cfs in enumerate(runoff_results_cfs, start=1):
    # Print one concise line per subcatchment.
    print(f"Subcatchment {index}: {runoff_cfs:.2f} cfs")

# Print accumulated design summary values.
print(f"Total runoff: {total_runoff_cfs:.2f} cfs")

# Print maximum value and checked count.
print(f"Maximum runoff: {maximum_runoff_cfs:.2f} cfs")
print(f"Subcatchments checked: {subcatchment_count}")

```

## **2. Controlled While Loops**

### **2.1. Trial Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780892995" width="250">



>* Test assumed values until acceptable
>* Revise trials based on design checks

>* Start with a realistic trial value
>* Check results and adjust each trial

>* Repeat trials with clear control
>* Justify safe, practical design decisions



```
#@title Python Code - Trial Calculations

# Controlled loops guide engineering trial calculations.
# Each trial checks one design assumption.
# Clear stopping conditions prevent endless repetition.

# Set required and initial pipe flow values.
required_flow_lps = 180
trial_diameter_mm = 300
step_size_mm = 25

# Store loop safety and trial tracking values.
maximum_diameter_mm = 600
trial_count = 0
capacity_lps = 0

# Repeat while capacity remains below requirement.
while capacity_lps < required_flow_lps and trial_diameter_mm <= maximum_diameter_mm:
    # Count each engineering trial calculation.
    trial_count = trial_count + 1
    # Estimate capacity using a simplified diameter relationship.
    capacity_lps = 0.00065 * trial_diameter_mm ** 2

    # Print a compact trace of each trial.
    print("Trial", trial_count, "diameter", trial_diameter_mm, "mm capacity", round(capacity_lps, 1), "L/s")
    # Increase diameter only when another trial is needed.
    if capacity_lps < required_flow_lps:
        trial_diameter_mm = trial_diameter_mm + step_size_mm

# Report whether the loop found an acceptable design.
if capacity_lps >= required_flow_lps:
    print("Accepted diameter:", trial_diameter_mm, "mm after", trial_count, "trials")
else:
    print("No acceptable diameter within the selected search limit")

```

### **2.2. Stopping Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780892997" width="250">



>* Stop when the engineering goal is met
>* Avoid endless, meaningless repeated calculations

>* Balance stopping rules to avoid unsafe results
>* Use tolerances and limits for reviewable loops

>* Use success and safety stopping conditions
>* Stop when trials become impractical



```
#@title Python Code - Stopping Conditions

# Demonstrate while loops with stopping conditions.
# Use a footing width trial calculation.
# Include success and safety stop rules.

# Define the required allowable bearing pressure.
required_pressure_kpa = 180.0

# Start with a small trial footing width.
footing_width_m = 0.80

# Set load and footing length assumptions.
column_load_kn = 900.0
footing_length_m = 2.50

# Set practical trial and increment limits.
width_step_m = 0.10
maximum_width_m = 3.00
maximum_trials = 30

# Track the number of trial calculations.
trial_count = 0
bearing_pressure_kpa = column_load_kn / (footing_width_m * footing_length_m)

# Continue while both safety checks still fail.
while bearing_pressure_kpa > required_pressure_kpa and trial_count < maximum_trials:
    trial_count = trial_count + 1
    footing_width_m = footing_width_m + width_step_m

    # Recalculate pressure after changing width.
    bearing_pressure_kpa = column_load_kn / (footing_width_m * footing_length_m)

    # Stop if the footing becomes unrealistic.
    if footing_width_m > maximum_width_m:
        break

# Decide why the loop stopped.
passed_pressure = bearing_pressure_kpa <= required_pressure_kpa
within_width_limit = footing_width_m <= maximum_width_m

# Print a short engineering summary.
print("Controlled while loop: footing width trial")
print(f"Trials completed: {trial_count}")
print(f"Final width: {footing_width_m:.2f} m")
print(f"Final pressure: {bearing_pressure_kpa:.1f} kPa")

# Report the stopping condition clearly.
if passed_pressure and within_width_limit:
    print("Stopped because bearing pressure is acceptable.")
else:
    print("Stopped because a safety limit was reached.")

```

### **2.3. Convergence Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780893002" width="250">



>* Check when trial estimates are close enough
>* Avoid unreliable results or wasted repetition

>* Match convergence tolerance to engineering context
>* Avoid overly loose or strict stopping limits

>* Detect failed or unstable trial calculations
>* Limit attempts and define success clearly



```
#@title Python Code - Convergence Checks

# Controlled while loops check convergence.
# Engineering tolerance decides acceptable closeness.
# Safety limits prevent endless trials.

# Set the target basin storage volume.
target_storage_m3 = 500.0

# Choose an initial trial storage estimate.
storage_m3 = 120.0

# Define tolerance and maximum trials.
tolerance_m3 = 2.0
max_trials = 20
trial_count = 0

# Save the first error magnitude.
previous_error_m3 = abs(target_storage_m3 - storage_m3)

# Repeat while error remains too large.
while previous_error_m3 > tolerance_m3 and trial_count < max_trials:
    trial_count = trial_count + 1
    correction_m3 = 0.45 * (target_storage_m3 - storage_m3)

    storage_m3 = storage_m3 + correction_m3
    current_error_m3 = abs(target_storage_m3 - storage_m3)
    previous_error_m3 = current_error_m3

# Check whether convergence was achieved.
converged = previous_error_m3 <= tolerance_m3

# Report the controlled iteration result.
print("Convergence check for basin storage.")
print(f"Final estimate: {storage_m3:.1f} m^3")
print(f"Target storage: {target_storage_m3:.1f} m^3")
print(f"Final error: {previous_error_m3:.2f} m^3")
print(f"Trials used: {trial_count} of {max_trials}")
print(f"Converged within tolerance: {converged}")

```

## **3. Accumulation**

### **3.1. Running Totals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780893003" width="250">



>* Start at zero, add each new value
>* Combine many contributions into one engineering total

>* Update totals as field data arrives
>* Trace results to catch possible errors

>* Add only compatible, consistent engineering quantities
>* Start clearly to produce reliable project totals



```
#@title Python Code - Running Totals

# Running totals combine repeated engineering contributions.
# Each loop update preserves accumulated progress.
# Printed steps help check calculation reasonableness.

# Store pavement segment lengths in meters.
segment_lengths_m = [120, 85, 140, 95]

# Store pavement width and asphalt thickness.
road_width_m = 7.5
asphalt_thickness_m = 0.08

# Start with no asphalt volume accumulated.
total_volume_m3 = 0.0

# Print a compact table heading.
print("Segment | Volume m^3 | Running total m^3")

# Add each segment contribution into total.
for segment_number, length_m in enumerate(segment_lengths_m, start=1):
    segment_volume_m3 = length_m * road_width_m * asphalt_thickness_m
    total_volume_m3 = total_volume_m3 + segment_volume_m3
    print(f"{segment_number:7d} | {segment_volume_m3:10.1f} | {total_volume_m3:17.1f}")

# Count the number of processed segments.
segment_count = len(segment_lengths_m)

# Compute a useful average contribution.
average_volume_m3 = total_volume_m3 / segment_count

# Print concise project summary values.
print(f"Total asphalt volume: {total_volume_m3:.1f} cubic meters")
print(f"Average per segment: {average_volume_m3:.1f} cubic meters")

```

### **3.2. Maximum Tracking**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780893005" width="250">



>* Find the largest engineering result
>* Update the maximum during repeated calculations

>* Start with a sensible current maximum
>* Compare consistent values and update when larger

>* Record where or when maximums occur
>* Update related details with each new maximum



```
#@title Python Code - Maximum Tracking

# Track maximum values during repeated checks.
# Civil engineers often seek controlling cases.
# This example uses beam moment data.

# Store stations in feet along a beam.
stations_ft = [0, 5, 10, 15, 20, 25]

# Store computed bending moments consistently.
moments_kip_ft = [0, 38, 64, 71, 55, 0]

# Validate paired engineering data lengths.
if len(stations_ft) != len(moments_kip_ft):
    raise ValueError("Station and moment lists must match")

# Initialize maximum using the first result.
max_moment = moments_kip_ft[0]
max_station = stations_ft[0]

# Check each station and update maximum.
for station, moment in zip(stations_ft, moments_kip_ft):
    if moment > max_moment:
        max_moment = moment
        max_station = station

# Count stations exceeding a design threshold.
design_limit = 60
exceed_count = 0

# Accumulate exceedance count across results.
for moment in moments_kip_ft:
    if moment > design_limit:
        exceed_count = exceed_count + 1

# Report the controlling engineering case.
print("Maximum tracking for beam moment")
print(f"Maximum moment: {max_moment} kip-ft")
print(f"Critical station: {max_station} ft")
print(f"Stations above {design_limit} kip-ft: {exceed_count}")

```

### **3.3. Iteration Counts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780893007" width="250">



>* Counts track repeated engineering observations
>* Final counts summarize the full dataset

>* Count using a clear increase rule
>* Separate checked cases from problem cases

>* Counts verify all items were processed
>* Counts add context to engineering summaries



```
#@title Python Code - Iteration Counts

# Count iterations in engineering checks.
# Track total and concerning cases.
# Compare counted results with expectations.

# Small pavement rut depths, in millimeters.
rut_depths_mm = [5, 11, 8, 14, 6, 17]
limit_mm = 10

# Start both counters at zero.
sections_checked = 0
sections_over_limit = 0
maximum_rut_mm = 0

# Update counts during each iteration.
for rut_depth_mm in rut_depths_mm:
    sections_checked = sections_checked + 1
    maximum_rut_mm = max(maximum_rut_mm, rut_depth_mm)

    # Count only sections needing attention.
    if rut_depth_mm > limit_mm:
        sections_over_limit = sections_over_limit + 1

# Confirm expected number of sections.
expected_sections = len(rut_depths_mm)
count_matches_expected = sections_checked == expected_sections

# Compute a simple percentage summary.
percent_over_limit = sections_over_limit / sections_checked * 100

# Print a compact engineering summary.
print("Iteration count summary for pavement rutting")
print("Sections checked:", sections_checked)
print("Sections over limit:", sections_over_limit)
print("Percent over limit:", round(percent_over_limit, 1))
print("Maximum rut depth, mm:", maximum_rut_mm)
print("Count matches expected:", count_matches_expected)

```
