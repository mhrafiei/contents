## **1. Boolean Logic**

### **1.1. Compare Limits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780881772" width="250">



>* Boolean checks compare results with limits
>* Limits guide safe engineering decisions

>* Identify whether limits are maximums or minimums
>* Compare values in the correct direction

>* Use clear wording and matching units
>* Define boundary equality for defensible checks



```
#@title Python Code - Compare Limits

# Compare calculated values with engineering limits.
# Boolean results guide simple engineering decisions.
# Clear messages support safe review.

# Define calculated engineering values.
deflection_mm = 18.5
settlement_mm = 27.0

# Define allowable engineering limits.
max_deflection_mm = 20.0
max_settlement_mm = 25.0

# Compare results with maximum limits.
deflection_ok = deflection_mm <= max_deflection_mm
settlement_ok = settlement_mm <= max_settlement_mm

# Print compact Boolean check results.
print("Deflection within limit:", deflection_ok)
print("Settlement within limit:", settlement_ok)

# Classify beam serviceability condition.
if deflection_ok and settlement_ok:
    message = "Both serviceability checks are acceptable."
elif deflection_ok or settlement_ok:
    message = "Warning: one serviceability check exceeds its limit."
else:
    message = "Warning: both serviceability checks exceed their limits."

# Print the engineering decision message.
print(message)

# Compute remaining margin for deflection.
deflection_margin_mm = max_deflection_mm - deflection_mm
settlement_margin_mm = max_settlement_mm - settlement_mm

# Print margins using consistent units.
print("Deflection margin in mm:", deflection_margin_mm)
print("Settlement margin in mm:", settlement_margin_mm)

```

### **1.2. Engineering Limit Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780881774" width="250">



>* Compare calculated results with engineering limits
>* Use true-false checks for consistent decisions

>* State comparison direction clearly
>* Check maximums, minimums, or acceptable ranges

>* Check equality rules, tolerances, and units
>* Use Boolean checks for sound decisions



```
#@title Python Code - Engineering Limit Checks

# Boolean checks compare results with limits.
# Engineering decisions often become true or false.
# Clear messages help reviewers understand failures.

# Store calculated engineering values and allowable limits.
deflection_in = 0.82
limit_deflection_in = 0.75

# Build a maximum limit Boolean expression.
deflection_passes = deflection_in <= limit_deflection_in
print("Deflection within limit:", deflection_passes)

# Store bearing pressure and allowable capacity.
bearing_kpa = 185.0
allowable_kpa = 200.0

# Build another clear maximum limit check.
bearing_passes = bearing_kpa <= allowable_kpa
print("Bearing pressure within limit:", bearing_passes)

# Store factor of safety and minimum requirement.
factor_of_safety = 1.45
minimum_factor = 1.50

# Build a minimum limit Boolean expression.
safety_passes = factor_of_safety >= minimum_factor
print("Factor of safety acceptable:", safety_passes)

# Classify results using conditional logic.
if deflection_passes and bearing_passes and safety_passes:
    status_message = "All engineering checks pass."
elif not safety_passes:
    status_message = "Warning: factor of safety is too low."
else:
    status_message = "Warning: one serviceability limit is exceeded."

# Print one final engineering warning message.
print(status_message)

```

### **1.3. Threshold Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780881776" width="250">



>* Turn engineering limits into true-or-false checks
>* Compare results clearly against acceptable thresholds

>* Match comparison direction to engineering risk
>* Check for limits or acceptable ranges

>* Match units, assumptions, and design context
>* Use checks to support clear decisions



```
#@title Python Code - Threshold Checks

# Threshold checks compare results with limits.
# Boolean expressions return True or False.
# Civil engineers use them for decisions.

# Store a calculated beam deflection value.
deflection_mm = 18.5

# Store the allowable serviceability limit.
allowable_deflection_mm = 20.0

# Compare deflection against the engineering limit.
deflection_is_acceptable = deflection_mm <= allowable_deflection_mm

# Store a calculated culvert flow value.
flow_cfs = 125.0

# Store the culvert design capacity.
capacity_cfs = 110.0

# Compare flow against hydraulic capacity.
flow_exceeds_capacity = flow_cfs > capacity_cfs

# Store a calculated factor of safety.
factor_of_safety = 1.42

# Store the required minimum value.
minimum_factor_of_safety = 1.50

# Compare safety factor against requirement.
safety_factor_is_low = factor_of_safety < minimum_factor_of_safety

# Print concise threshold check results.
print("Beam deflection acceptable:", deflection_is_acceptable)
print("Culvert flow exceeds capacity:", flow_exceeds_capacity)
print("Factor of safety below minimum:", safety_factor_is_low)

# Use conditions to create engineering warnings.
if not deflection_is_acceptable:
    print("Warning: beam deflection exceeds allowable limit.")
elif flow_exceeds_capacity:
    print("Warning: culvert flow exceeds design capacity.")
elif safety_factor_is_low:
    print("Warning: factor of safety is below requirement.")
else:
    print("All checked engineering thresholds are acceptable.")

```

## **2. Branching Rules**

### **2.1. If Statements**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780881760" width="250">



>* If statements run actions when conditions pass
>* Engineering limits can trigger specific responses

>* Write conditions with engineering meaning
>* Turn standards into repeatable decisions

>* Clear conditions make classifications defensible
>* Consistent logic automates routine engineering decisions



```
#@title Python Code - If Statements

# Demonstrate if statements for engineering checks.
# Compare calculated values with clear limits.
# Print concise classifications for learners.

# Store one settlement result in millimeters.
settlement_mm = 32.0

# Store the serviceability limit in millimeters.
settlement_limit_mm = 25.0

# Check whether settlement exceeds the limit.
if settlement_mm > settlement_limit_mm:
    status = "Warning: settlement exceeds the serviceability limit."

# Print the measured value and classification.
print(f"Settlement: {settlement_mm} mm")
print(status)

# Store one channel water level in meters.
water_level_m = 2.4

# Store the alert level in meters.
alert_level_m = 2.0

# Check whether water level needs review.
if water_level_m > alert_level_m:
    channel_message = "Warning: channel water level requires review."

# Print the measured value and classification.
print(f"Water level: {water_level_m} m")
print(channel_message)

```

### **2.2. Else Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780881764" width="250">



>* Else handles cases after earlier checks fail
>* It makes classifications complete and predictable

>* Else depends on earlier condition order
>* Poor checks can misclassify engineering results

>* Use else for normal default outcomes
>* Check invalid data before final classification



```
#@title Python Code - Else Conditions

# Else branches complete engineering classifications.
# Earlier checks handle risky cases.
# Remaining cases reach final else.

# Define slope factors of safety.
slope_names = ["North Cut", "East Fill", "West Bench"]

# Store calculated factors of safety.
factors_of_safety = [0.95, 1.18, 1.42]

# Classify each slope condition.
for slope_name, safety_factor in zip(slope_names, factors_of_safety):

    # Check the most severe case first.
    if safety_factor < 1.00:
        status = "UNSAFE: immediate engineering review required"

    # Check the borderline monitoring range.
    elif safety_factor < 1.30:
        status = "MONITOR: add inspections after rainfall"

    # Use else for all remaining acceptable values.
    else:
        status = "ACCEPTABLE: factor of safety meets criterion"

    # Print one clear classification message.
    print(slope_name, "FS =", safety_factor, "->", status)

# Summarize the lesson about else.
print("The else branch handled values not caught earlier.")

```

### **2.3. Else Conditions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780881762" width="250">



>* Else handles cases after all checks fail
>* Keeps classifications complete, clear, and reliable

>* Else catches cases after prior checks fail
>* Check serious conditions first for correct classification

>* Else branches classify every evaluated case
>* They confirm satisfactory results after warnings



```
#@title Python Code - Else Conditions

# Else branches complete engineering classifications.
# Earlier checks handle worse conditions first.
# Remaining cases reach the safe branch.

# Define pavement distress values in percent.
cracking_percent = 7.5
rut_depth_inches = 0.18

# Classify using serious checks before else.
if cracking_percent > 25 or rut_depth_inches > 0.75:
    condition = "Severe distress: schedule rehabilitation."
elif cracking_percent > 12 or rut_depth_inches > 0.40:
    condition = "Moderate distress: plan maintenance soon."
elif cracking_percent > 5 or rut_depth_inches > 0.20:
    condition = "Minor distress: continue routine monitoring."
else:
    condition = "Acceptable condition: no warning required."

# Show the measurements used for classification.
print("Cracking percent:", cracking_percent)
print("Rut depth inches:", rut_depth_inches)

# Show the final branch selected.
print("Pavement classification:", condition)

```

## **3. Engineering Warning Criteria**

### **3.1. Threshold Warning Messages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780881782" width="250">



>* Show when results pass acceptable limits
>* Explain needed monitoring or review actions

>* Name the exceeded check and threshold
>* Explain the practical engineering concern clearly

>* Match warning tone to severity level
>* Suggest measured follow-up, not unnecessary alarm



```
#@title Python Code - Threshold Warning Messages

# Threshold warnings connect numbers to decisions.
# This example checks civil engineering limits.
# Messages stay firm without sounding critical.

# Store calculated results and warning thresholds.
checks = [
    ("Pavement deflection", 0.42, 0.35, "inches"),
    ("Retaining wall movement", 18.0, 15.0, "millimeters"),
    ("Channel flow depth", 3.1, 3.5, "feet"),
]

# Start a list for concise status messages.
messages = []

# Classify each condition against its threshold.
for name, value, limit, unit in checks:
    if value > limit:
        message = f"WARNING: {name} exceeds {limit} {unit}; review soon."
    else:
        message = f"OK: {name} remains within {limit} {unit}."

    messages.append(message)

# Print each warning or normal message.
for message in messages:
    print(message)

# Summarize how many checks need attention.
warning_count = sum(message.startswith("WARNING") for message in messages)

# Print a short management summary.
print(f"Checks needing follow-up: {warning_count} of {len(checks)}.")

```

### **3.2. Threshold Warning Messages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780881779" width="250">



>* Flag limit exceedances for engineering review
>* Explain condition, concern, and needed attention

>* Name the exceeded engineering limit clearly
>* Guide teams toward appropriate next actions

>* Use serious but not alarming warning language
>* Link exceedances to concerns and next steps



```
#@title Python Code - Threshold Warning Messages

# Threshold messages connect calculations to engineering decisions.
# Limits help classify results without exaggerating risk.
# Clear warnings state condition, criterion, and response.

# Store beam deflection data in inches.
span_inches = 240
actual_deflection_inches = 0.92
allowable_deflection_inches = span_inches / 360

# Compare the calculated value with the limit.
excess_inches = actual_deflection_inches - allowable_deflection_inches
status_label = "Review needed"
warning_message = ""

# Use conditional logic to create a useful message.
if actual_deflection_inches <= allowable_deflection_inches:
    status_label = "Acceptable"
    warning_message = "Deflection is within the serviceability limit."
elif excess_inches <= 0.25:
    status_label = "Threshold warning"
    warning_message = "Deflection exceeds serviceability limit; review performance and comfort."
else:
    status_label = "High warning"
    warning_message = "Deflection greatly exceeds limit; request senior engineering review."

# Print a concise engineering check summary.
print("Engineering Warning Criteria: Threshold Warning Messages")
print(f"Span: {span_inches} inches")
print(f"Allowable deflection: {allowable_deflection_inches:.2f} inches")
print(f"Measured deflection: {actual_deflection_inches:.2f} inches")

# Print the classification and recommended attention.
print(f"Classification: {status_label}")
print(f"Message: {warning_message}")
print("Action: Document result and schedule appropriate follow-up.")

```

### **3.3. Critical Warning Messages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780881780" width="250">



>* Flag unsafe results beyond serious engineering limits
>* Prompt urgent review, action, or shutdown

>* Name the failed condition and limit
>* Explain why the result matters

>* Use direct, professional warning language
>* Link actions to engineering consequences



```
#@title Python Code - Critical Warning Messages

# This example builds critical engineering warning messages.
# Each check compares results against serious limits.
# Clear wording explains the exceeded engineering criterion.

# Store calculated values and engineering criteria.
beam_stress_mpa = 285.0
allowable_stress_mpa = 250.0

# Store slope stability values for checking.
slope_factor = 1.05
minimum_factor = 1.30

# Store roadway flood elevations for checking.
flood_level_ft = 13.8
road_protection_ft = 12.5

# Check beam stress against allowable capacity.
if beam_stress_mpa > allowable_stress_mpa:
    beam_message = "CRITICAL: Beam stress exceeds allowable capacity; engineering review required."
elif beam_stress_mpa == allowable_stress_mpa:
    beam_message = "NOTICE: Beam stress equals allowable capacity; verify assumptions carefully."
else:
    beam_message = "OK: Beam stress remains below allowable capacity."

# Check slope factor against required margin.
if slope_factor < minimum_factor:
    slope_message = "CRITICAL: Slope stability factor is below minimum safety margin."
elif slope_factor == minimum_factor:
    slope_message = "NOTICE: Slope stability factor exactly meets the minimum requirement."
else:
    slope_message = "OK: Slope stability factor exceeds the minimum requirement."

# Check flood elevation against protected roadway level.
if flood_level_ft > road_protection_ft:
    flood_message = "CRITICAL: Floodwater exceeds roadway protection level; closure review needed."
elif flood_level_ft == road_protection_ft:
    flood_message = "NOTICE: Floodwater reaches the roadway protection level."
else:
    flood_message = "OK: Floodwater remains below the protected roadway level."

# Print concise warning results for learners.
print("Engineering Warning Criteria")
print(beam_message)
print(slope_message)
print(flood_message)

```
