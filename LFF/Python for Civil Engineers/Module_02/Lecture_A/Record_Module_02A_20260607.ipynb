## **1. Boolean Logic**

### **1.1. Compare Engineering Limits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780890388" width="250">



>* Compare results with allowable engineering limits
>* Turn calculations into true-or-false decisions

>* Match comparisons to limit meaning
>* Use <= for maximums, >= for minimums

>* Use compatible units and sensible precision
>* Review small margins with engineering judgment



```
#@title Python Code - Compare Engineering Limits

# Compare calculated values with engineering limits.
# Boolean expressions return true or false.
# Units must match before comparisons.

# Define calculated values and limits.
beam_deflection_mm = 18.0
allowable_deflection_mm = 20.0

# Check a maximum serviceability limit.
deflection_ok = beam_deflection_mm <= allowable_deflection_mm

# Define a minimum safety requirement.
slope_factor_safety = 1.35
required_factor_safety = 1.50

# Check a minimum stability limit.
safety_ok = slope_factor_safety >= required_factor_safety

# Define an imperial clearance example.
bridge_clearance_ft = 14.2
required_clearance_ft = 14.0

# Check another minimum requirement.
clearance_ok = bridge_clearance_ft >= required_clearance_ft

# Print compact engineering decisions.
print("Deflection passes maximum limit:", deflection_ok)
print("Slope safety passes minimum limit:", safety_ok)
print("Bridge clearance passes minimum limit:", clearance_ok)

# Create clear warning messages.
if not deflection_ok:
    print("Warning: beam deflection exceeds allowable limit.")

if not safety_ok:
    print("Warning: slope factor of safety is too low.")

if not clearance_ok:
    print("Warning: bridge clearance is below requirement.")

# Summarize overall design acceptability.
all_checks_ok = deflection_ok and safety_ok and clearance_ok
print("All engineering limit checks pass:", all_checks_ok)

```

### **1.2. Compare Engineering Limits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780890390" width="250">



>* Compare calculated results with engineering limits
>* Boolean checks support engineering judgment

>* Match comparisons to engineering intent
>* Check upper, lower, or range limits

>* Limits need context and precise logic
>* Clear checks flag problems early



```
#@title Python Code - Compare Engineering Limits

# Boolean checks compare results with limits.
# Engineering limits create true false decisions.
# Clear comparisons support safe design judgment.

# Store calculated and allowable beam deflections.
beam_deflection_mm = 18.0
allowable_deflection_mm = 20.0

# Check an upper limit using less than equal.
deflection_within_limit = beam_deflection_mm <= allowable_deflection_mm
print("Beam deflection acceptable:", deflection_within_limit)

# Store calculated and required safety factors.
slope_factor_safety = 1.42
required_factor_safety = 1.50

# Check a lower limit using greater than equal.
slope_meets_requirement = slope_factor_safety >= required_factor_safety
print("Slope safety factor acceptable:", slope_meets_requirement)

# Store estimated and allowable foundation settlements.
settlement_in = 1.25
allowable_settlement_in = 1.00

# Check an exceedance using greater than.
settlement_exceeds_limit = settlement_in > allowable_settlement_in
print("Foundation settlement exceeds limit:", settlement_exceeds_limit)

# Store measured pipe velocity range limits.
pipe_velocity_mps = 2.4
minimum_velocity_mps = 0.6
maximum_velocity_mps = 3.0

# Check an acceptable range using and.
velocity_in_range = minimum_velocity_mps <= pipe_velocity_mps <= maximum_velocity_mps
print("Pipe velocity within range:", velocity_in_range)

# Summarize one practical warning condition.
warning_needed = settlement_exceeds_limit or not slope_meets_requirement
print("Engineering warning needed:", warning_needed)

```

### **1.3. Threshold Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780890392" width="250">



>* Compare calculated results with acceptable limits
>* Identify when engineering boundaries are crossed

>* Match comparison direction to the limit type
>* Make threshold logic clear and unambiguous

>* Handle near-limit values with engineering judgment
>* Use warnings to support clear decisions



```
#@title Python Code - Threshold Checks

# Compare engineering values with limits.
# Boolean checks return true or false.
# Clear messages support design review.

# Store calculated values and limits.
beam_deflection_mm = 18.0
allowable_deflection_mm = 20.0
pipe_velocity_ft_s = 1.8

# Store minimum acceptable velocity limit.
minimum_velocity_ft_s = 2.0
factor_of_safety = 1.65
required_factor_of_safety = 1.50

# Build Boolean threshold checks.
deflection_ok = beam_deflection_mm <= allowable_deflection_mm
velocity_ok = pipe_velocity_ft_s >= minimum_velocity_ft_s
safety_ok = factor_of_safety >= required_factor_of_safety

# Print concise results for learners.
print("Threshold checks for civil engineering values:")
print(f"Beam deflection acceptable: {deflection_ok}")
print(f"Pipe velocity acceptable: {velocity_ok}")
print(f"Factor of safety acceptable: {safety_ok}")

# Create clear warning messages.
if not deflection_ok:
    print("Warning: beam deflection exceeds allowable limit.")
elif not velocity_ok:
    print("Warning: pipe velocity is below minimum limit.")
else:
    print("Primary checks passed, review safety separately.")

# Check safety after serviceability messages.
if safety_ok:
    status_message = "Factor of safety meets the requirement."
else:
    status_message = "Warning: factor of safety is too low."

# Print the final safety status.
print(status_message)

```

## **2. Branching Rules**

### **2.1. If Statement Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780890394" width="250">



>* If statements test true-or-false conditions
>* Results guide engineering classification and action

>* If statements create controlled program branches
>* Conditions trigger engineering classifications or actions

>* Use precise engineering limits for conditions
>* Choose specific actions for consistent judgments



```
#@title Python Code - If Statement Basics

# This example introduces one simple engineering decision.
# It compares rut depth against a maintenance limit.
# The if block runs only when true.

# Store a measured pavement rut depth.
rut_depth_inches = 0.72
allowable_rut_inches = 0.50

# Display the basic check being performed.
print("Pavement rut depth check")
print(f"Measured rut: {rut_depth_inches:.2f} inches")

# Test whether the engineering limit is exceeded.
if rut_depth_inches > allowable_rut_inches:
    print("Warning: rut depth exceeds maintenance threshold.")

# Continue after the decision branch.
print("If statement check is complete.")

```

### **2.2. Else Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780890396" width="250">



>* Else handles cases when conditions are false
>* Every result gets a clear classification

>* Else completes engineering decision logic
>* Default outcomes clarify absent risk conditions

>* Write earlier conditions with care
>* Use else after specific categories



```
#@title Python Code - Else Classification

# Else branches complete engineering decisions.
# This example classifies pavement deflection.
# The output stays short and clear.

# Store measured and allowable deflection values.
measured_deflection_mm = 7.4
allowable_deflection_mm = 8.0

# Compare the calculated value against the limit.
exceeds_limit = measured_deflection_mm > allowable_deflection_mm

# Use else for the remaining classification.
if exceeds_limit:
    pavement_status = "Warning: deflection exceeds allowable limit."
else:
    pavement_status = "Acceptable: deflection is within allowable limit."

# Print the engineering inputs and result.
print("Measured deflection:", measured_deflection_mm, "mm")
print("Allowable deflection:", allowable_deflection_mm, "mm")
print("Classification:", pavement_status)

```

### **2.3. Elif Else Paths**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780890398" width="250">



>* Elif handles multiple ordered engineering categories
>* Clear logic supports reliable safety decisions

>* Check stricter conditions before broader ones
>* Order reflects engineering priority and risk

>* Else handles default or unexpected cases
>* Every input receives a clear outcome



```
#@title Python Code - Elif Else Paths

# Classify stormwater rainfall using ordered branches.
# Elif paths choose the first true condition.
# Clear messages support engineering decisions.

# Define measured rainfall intensity.
rainfall_mm_per_hr = 72

# Define design review limits.
watch_limit = 25
warning_limit = 50
closure_limit = 75

# Print the measured value.
print("Rainfall intensity:", rainfall_mm_per_hr, "mm/hr")

# Check the most severe case first.
if rainfall_mm_per_hr >= closure_limit:
    status = "Immediate closure review required"

elif rainfall_mm_per_hr >= warning_limit:
    status = "Urgent drainage inspection required"

elif rainfall_mm_per_hr >= watch_limit:
    status = "Routine monitoring recommended"

else:
    status = "No stormwater action required"

# Print the selected branch result.
print("Classification:", status)

# Explain why later branches are skipped.
print("Python used the first true branch only.")

```

## **3. Engineering Warning Criteria**

### **3.1. Warning Messages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780890383" width="250">



>* Warnings explain limit exceedances clearly
>* Vague warnings delay safe engineering responses

>* Tailor warnings to each user's needs
>* Show severity so responses match risk

>* State exceedance details for team accountability
>* Keep warnings concise, calm, and actionable



```
#@title Python Code - Warning Messages

# Clear warnings help engineers respond quickly.
# Conditional checks compare results with limits.
# Messages should explain the exceeded criterion.

# Store a calculated beam deflection value.
beam_deflection_in = 1.15

# Store the serviceability deflection limit.
deflection_limit_in = 0.90

# Store the location being checked.
member_name = "Beam B12"

# Compare the result against the limit.
if beam_deflection_in > deflection_limit_in:

    # Build a clear engineering warning.
    message = (
        f"WARNING: {member_name} deflection is "
        f"{beam_deflection_in:.2f} in, exceeding "

        f"the {deflection_limit_in:.2f} in serviceability limit."
    )

# Handle results that remain acceptable.
else:

    # Build a calm acceptable message.
    message = (
        f"OK: {member_name} deflection is "
        f"{beam_deflection_in:.2f} in, within "

        f"the {deflection_limit_in:.2f} in serviceability limit."
    )

# Print the final warning message.
print(message)

```

### **3.2. Limit Exceedance Alerts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780890385" width="250">



>* Flag results beyond engineering criteria
>* Explain consequences for appropriate response

>* Name the exceeded limit and triggering result
>* Explain the condition and needed response

>* Use clear alerts for quick decisions
>* Match urgency to engineering risk



```
#@title Python Code - Limit Exceedance Alerts

# Conditional alerts communicate engineering limit exceedances.
# Clear warnings include value, limit, and consequence.
# This example checks several civil measurements.

# Store small engineering results and allowable limits.
checks = [
    ("Beam deflection", 1.35, 1.00, "in", "serviceability concern"),
    ("Wall movement", 18.0, 12.0, "mm", "stability review needed"),
    ("Channel depth", 4.2, 4.5, "ft", "within monitored range"),
]

# Print a compact heading for the alerts.
print("Limit Exceedance Alerts")

# Compare each calculated result with its limit.
for name, value, limit, unit, consequence in checks:
    if value > limit:
        status = "WARNING"
    elif value == limit:
        status = "CAUTION"
    else:
        status = "OK"

    # Build a clear message for the reader.
    message = (
        f"{status}: {name} = {value} {unit}; "
        f"limit = {limit} {unit}; {consequence}."
    )

    # Show one concise alert line.
    print(message)

# Count results that exceed their criteria.
exceedances = sum(value > limit for _, value, limit, _, _ in checks)

# Summarize the required engineering attention.
summary = f"Total exceedance alerts requiring attention: {exceedances}."
print(summary)

```

### **3.3. Clear Warning Messages**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780890386" width="250">



>* Explain the detected engineering problem clearly
>* Identify impacts, affected elements, and needed action

>* State the issue, value, limit, and location
>* Make warnings specific enough to guide action

>* Keep warnings calm, direct, and serious
>* Use consistent wording to guide decisions



```
#@title Python Code - Clear Warning Messages

# Clear messages connect checks to engineering action.
# This example compares values with stated limits.
# Each warning names condition, value, and response.

# Store calculated civil engineering results.
bridge_deflection_mm = 38.0
allowable_deflection_mm = 25.0

# Store a second field inspection result.
wall_safety_factor = 1.25
required_safety_factor = 1.50

# Check bridge deflection against serviceability limit.
if bridge_deflection_mm > allowable_deflection_mm:
    print(f"WARNING: Bridge girder deflection is {bridge_deflection_mm:.1f} mm; limit is {allowable_deflection_mm:.1f} mm. Engineering review is required.")
else:
    print("Bridge girder deflection is within the allowable limit.")

# Check retaining wall safety against minimum requirement.
if wall_safety_factor < required_safety_factor:
    print(f"WARNING: Retaining wall factor of safety is {wall_safety_factor:.2f}; required minimum is {required_safety_factor:.2f}. Review stability before proceeding.")
else:
    print("Retaining wall factor of safety meets the requirement.")

# Summarize why clear messages are useful.
print("Clear warnings identify the component, result, limit, and needed attention.")

```
