## **1. String Labels**

### **1.1. Project Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780880890" width="250">



>* Project labels give engineering data clear context
>* Clear labels make calculations traceable and useful

>* Labels help teams understand engineering analyses
>* Labels prevent confusion across similar measurements

>* Labels create professional, repeatable engineering reports
>* Context keeps results clear and trustworthy



```
#@title Python Code - Project Labels

# Strings make engineering outputs readable.
# Project labels describe calculation context.
# Clear labels reduce reporting confusion.

# Store identifying labels as string values.
project_name = "Riverbend Pedestrian Bridge"
engineer_name = "A. Patel"
report_type = "Preliminary Deflection Check"

# Store scenario labels for later output.
span_label = "Main span"
load_case = "Service live load"
material_label = "Reinforced concrete deck"

# Store small numeric engineering values.
span_ft = 48
estimated_deflection_in = 0.62
allowable_deflection_in = 0.80

# Combine labels and values in readable lines.
print("PROJECT LABELS REPORT")
print("Project:", project_name)
print("Engineer:", engineer_name)
print("Report type:", report_type)

# Print labels beside engineering quantities.
print("Location:", span_label)
print("Material:", material_label)
print("Load case:", load_case)

# Print calculated values with units.
print("Span length:", span_ft, "ft")
print("Estimated deflection:", estimated_deflection_in, "in")
print("Allowable deflection:", allowable_deflection_in, "in")

# Use labels to make the conclusion clear.
check_result = estimated_deflection_in <= allowable_deflection_in
status_label = "Passes limit" if check_result else "Exceeds limit"
print("Deflection status:", status_label)

```

## **2. User Input Basics**

### **2.1. Integer Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780880906" width="250">



>* Convert whole-number input from text to integers
>* Use counts correctly in engineering calculations

>* Numeric-looking input remains text until converted
>* Integer conversion enables correct engineering calculations

>* Validate integer entries for reasonable values
>* Use clear prompts to improve reliability



```
#@title Python Code - Integer Input

# Demonstrate integer input conversion.
# Use civil engineering count examples.
# Format results for reporting.

# Store simulated user entry text.
entry_text = "12"

# Convert the text into integer.
beam_count = int(entry_text)

# Store one beam length value.
length_per_beam_ft = 18

# Calculate total beam length.
total_length_ft = beam_count * length_per_beam_ft

# Print original text and type.
print("Typed entry:", entry_text, "is", type(entry_text).__name__)

# Print converted value and type.
print("Converted count:", beam_count, "is", type(beam_count).__name__)

# Print formatted engineering result.
print(f"Report: {beam_count} beams need {total_length_ft} ft total length.")

```

### **2.2. Float Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780880902" width="250">



>* Convert decimal input text into numbers
>* Use floats for engineering calculations

>* Keep decimal measurements accurate
>* Convert text for engineering calculations

>* Handle messy numeric entries carefully
>* Use clear prompts for reliable calculations



```
#@title Python Code - Float Conversion

# Float conversion prepares text measurements.
# Engineering inputs often include decimals.
# Converted floats support reliable calculations.

# Store simulated user entries as text.
span_text = "4.5"
load_text = "12.75"

# Convert text values into floating numbers.
span_meters = float(span_text)
load_kn_per_meter = float(load_text)

# Calculate total line load on beam.
total_load_kn = span_meters * load_kn_per_meter

# Format values for engineering-style reporting.
span_report = f"Beam span: {span_meters:.2f} m"
load_report = f"Uniform load: {load_kn_per_meter:.2f} kN/m"
result_report = f"Total load: {total_load_kn:.2f} kN"

# Display concise report lines for learners.
print("Float conversion example")
print(span_report)
print(load_report)
print(result_report)

```

### **2.3. Prompting For Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780880904" width="250">



>* Clear prompts specify values, units, and precision
>* Specific wording supports accurate numeric conversion

>* Write prompts users can easily answer
>* Distinguish counts from decimal measurements

>* Context builds confidence in entered values
>* Clear prompts improve conversions and calculations



```
#@title Python Code - Prompting For Values

# Clear prompts improve engineering input reliability.
# Simulated entries keep this notebook automatic.
# Numeric conversion supports later engineering calculations.

# Store realistic prompt text examples.
prompt_length = "Enter beam length in meters: "
prompt_load = "Enter design load in kilonewtons: "
prompt_bolts = "Enter number of anchor bolts: "

# Simulate typed responses from a user.
length_text = "6.5"
load_text = "42.0"
bolts_text = "8"

# Convert text entries into numeric values.
beam_length_m = float(length_text)
design_load_kn = float(load_text)
anchor_bolt_count = int(bolts_text)

# Compute simple engineering summary values.
load_per_meter_kn = design_load_kn / beam_length_m
load_per_bolt_kn = design_load_kn / anchor_bolt_count

# Print prompts alongside simulated typed answers.
print(prompt_length + length_text)
print(prompt_load + load_text)
print(prompt_bolts + bolts_text)

# Print formatted results for reporting.
print(f"Beam length: {beam_length_m:.2f} m")
print(f"Design load: {design_load_kn:.1f} kN")
print(f"Uniform load: {load_per_meter_kn:.2f} kN/m")
print(f"Load per bolt: {load_per_bolt_kn:.2f} kN/bolt")

```

## **3. Formatted Engineering Results**

### **3.1. Formatted Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780880892" width="250">



>* Turn raw calculations into readable results
>* Add labels and context for clarity

>* Combine labels with calculated values
>* Make results clear, traceable, and report-ready

>* Consistent formats make results easier to compare
>* Clear output supports repeatable engineering decisions



```
#@title Python Code - Formatted Values

# Format values for clear engineering reporting.
# Labels and units reduce result ambiguity.
# Rounded outputs support professional report readability.

# Store simple beam calculation inputs.
project_name = "Pedestrian Bridge Check"
span_m = 6.0
load_kN_per_m = 12.5

# Compute total load and support reaction.
total_load_kN = span_m * load_kN_per_m
reaction_kN = total_load_kN / 2

# Store a small deflection calculation result.
deflection_mm = 18.7364
allowable_mm = span_m * 1000 / 360

# Create formatted report lines.
line_one = f"Project: {project_name}"
line_two = f"Span length: {span_m:.2f} m"
line_three = f"Total uniform load: {total_load_kN:.1f} kN"
line_four = f"Reaction at each support: {reaction_kN:.2f} kN"

# Add decision wording with units.
line_five = f"Calculated deflection: {deflection_mm:.1f} mm"
line_six = f"Allowable deflection: {allowable_mm:.1f} mm"
status = "OK" if deflection_mm <= allowable_mm else "Review needed"

# Print a compact engineering result summary.
print(line_one)
print(line_two)
print(line_three)
print(line_four)

print(line_five)
print(line_six)
print(f"Deflection check status: {status}")

```

### **3.2. Decimal Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780880899" width="250">



>* Choose decimal places for readable results
>* Match displayed precision to engineering needs

>* Match decimals to measurement precision
>* Round results for clear engineering decisions

>* Consistent decimals make comparisons easier
>* Professional formatting builds trust in results



```
#@title Python Code - Decimal Formatting

# Decimal formatting supports engineering report readability.
# This example formats small civil results.
# Consistent decimals make comparisons easier.

# Store computed values from engineering calculations.
deflection_mm = 7.236847
stress_mpa = 182.98721
flow_lps = 14.5

# Format values for a concise report.
formatted_deflection = f"{deflection_mm:.2f} mm"
formatted_stress = f"{stress_mpa:.1f} MPa"
formatted_flow = f"{flow_lps:.2f} L/s"

# Show raw values before report formatting.
print("Raw deflection:", deflection_mm)
print("Raw stress:", stress_mpa)
print("Raw flow:", flow_lps)

# Show formatted values for engineering communication.
print("Report deflection:", formatted_deflection)
print("Report stress:", formatted_stress)
print("Report flow:", formatted_flow)

# Compare repeated values using aligned decimals.
span_a_cm = 243.456
span_b_cm = 243.4
print(f"Span A: {span_a_cm:8.2f} cm")
print(f"Span B: {span_b_cm:8.2f} cm")

```

### **3.3. Engineering Units**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780880901" width="250">



>* Units give numbers engineering meaning
>* Clear labels prevent confusion across teams

>* Units separate similar-looking engineering values
>* Clear labels prevent costly technical mistakes

>* Use consistent unit systems and labels
>* Place readable units beside every value



```
#@title Python Code - Engineering Units

# Engineering results need clear unit labels.
# Formatted strings make reports easier to read.
# This example combines metric and imperial units.

# Store project information as labeled strings.
project_name = "Campus walkway drainage check"
engineer_name = "Beginner Civil Engineer"

# Store calculated engineering values with units.
pipe_diameter_mm = 450.0
channel_length_ft = 128.5

# Convert numeric results into related units.
pipe_diameter_m = pipe_diameter_mm / 1000.0
channel_length_m = channel_length_ft * 0.3048

# Prepare formatted report lines with units.
line_one = f"Project: {project_name}"
line_two = f"Engineer: {engineer_name}"
line_three = f"Pipe diameter: {pipe_diameter_mm:.0f} mm ({pipe_diameter_m:.2f} m)"

# Prepare more formatted report lines.
line_four = f"Channel length: {channel_length_ft:.1f} ft ({channel_length_m:.2f} m)"
line_five = "Units are placed beside every value."

# Print a compact engineering report.
print(line_one)
print(line_two)
print(line_three)

# Print remaining formatted engineering results.
print(line_four)
print(line_five)

```
