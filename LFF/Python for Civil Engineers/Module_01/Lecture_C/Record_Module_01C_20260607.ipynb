## **1. String Basics**

### **1.1. Project Labels**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780889621" width="250">



>* Strings give numbers engineering context
>* Labels make outputs clear for others

>* Labels prevent engineering results from being misread
>* Descriptive strings make outputs confidently interpretable

>* Consistent labels support auditing and reuse
>* Clear strings reduce team communication confusion



```
#@title Python Code - Project Labels

# Strings give engineering values clear context.
# Project labels make outputs easier to review.
# This example labels a simple bridge case.

# Store project details as descriptive strings.
project_name = "Campus Pedestrian Bridge"
component_name = "Main steel girder"
design_case = "Preliminary service load check"

# Store measured or assumed numeric values.
span_length_ft = 42.0
service_load_kips = 18.5
allowable_load_kips = 25.0

# Build readable labels using f-strings.
project_label = f"Project: {project_name}"
component_label = f"Component: {component_name}"
case_label = f"Design case: {design_case}"

# Compute a simple engineering result.
load_ratio = service_load_kips / allowable_load_kips
percent_capacity = load_ratio * 100

# Print a short labeled engineering summary.
print(project_label)
print(component_label)
print(case_label)
print(f"Span length: {span_length_ft:.1f} ft")

# Label loads so their meaning is clear.
print(f"Service load: {service_load_kips:.1f} kips")
print(f"Allowable load: {allowable_load_kips:.1f} kips")
print(f"Capacity used: {percent_capacity:.1f}%")

# End with an interpretation label.
status = "acceptable" if load_ratio <= 1.0 else "over limit"
print(f"Check result: {status}")

```

## **2. User Input**

### **2.1. Integer Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780889629" width="250">



>* User input starts as text
>* Convert whole-number engineering counts to integers

>* Convert user responses before calculations
>* Check input format to avoid text errors

>* Check integer input for unreasonable responses
>* Use whole-number counts for reliable calculations



```
#@title Python Code - Integer Input

# Integer input converts text into whole numbers.
# Counts are common in civil engineering calculations.
# This example simulates typed user responses.

# Store sample responses as text strings.
beam_count_text = "5"
bolt_count_text = "8"

# Convert each text value into an integer.
beam_count = int(beam_count_text)
bolt_count = int(bolt_count_text)

# Use integers in a real arithmetic calculation.
total_bolts = beam_count * bolt_count

# Format the result for an engineering report.
report_line = f"Connection summary: {total_bolts} bolts are required."

# Print the original text and converted result.
print(f"Typed beam count was text: {beam_count_text!r}")
print(f"Converted beam count type: {type(beam_count).__name__}")
print(report_line)

```

### **2.2. Float Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780889627" width="250">



>* Decimal measurements require floating-point input
>* Convert text input for engineering calculations

>* Measured data often needs decimal precision
>* Whole-number input can distort engineering results

>* Prompt clearly for values and units
>* Convert decimals for accurate engineering calculations



```
#@title Python Code - Float Input

# Float input converts decimal text.
# Engineering prompts should include units.
# Converted floats support accurate calculations.

# Store simulated typed entries as text.
length_text = "2.75"
width_text = "1.40"
thickness_text = "0.18"

# Convert text entries into float values.
length_m = float(length_text)
width_m = float(width_text)
thickness_m = float(thickness_text)

# Compute a simple concrete slab volume.
area_m2 = length_m * width_m
volume_m3 = area_m2 * thickness_m

# Report inputs and formatted engineering result.
print("User length entry:", length_text, "m")
print("Converted length type:", type(length_m).__name__)
print("Slab plan area:", f"{area_m2:.2f} m^2")
print("Concrete volume:", f"{volume_m3:.3f} m^3")

```

### **2.3. Prompt Conversion Practice**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780889630" width="250">



>* User input starts as text
>* Convert text before doing calculations

>* Use integers for countable quantities
>* Use decimals for measured values

>* Clear prompts prevent input mistakes
>* Accurate conversions support reliable calculations



```
#@title Python Code - Prompt Conversion Practice

# Practice converting typed engineering values.
# Inputs arrive first as text.
# Converted numbers support correct calculations.

# Store simulated responses as text.
room_length_text = "4.8"
room_width_text = "3.2"
sample_count_text = "6"

# Convert measurements and counts deliberately.
room_length_m = float(room_length_text)
room_width_m = float(room_width_text)
sample_count = int(sample_count_text)

# Calculate area and average sample area.
room_area_m2 = room_length_m * room_width_m
area_per_sample_m2 = room_area_m2 / sample_count

# Format results for engineering reporting.
print(f"Length text became decimal: {room_length_m:.2f} m")
print(f"Width text became decimal: {room_width_m:.2f} m")
print(f"Sample count became integer: {sample_count} samples")
print(f"Room area: {room_area_m2:.2f} square meters")
print(f"Area per sample: {area_per_sample_m2:.2f} square meters")

```

## **3. Formatting Engineering Results**

### **3.1. Formatted Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780889620" width="250">



>* Raw numbers need context and units
>* Formatting makes results clear and usable

>* Clear formatting highlights important engineering results
>* Descriptive output supports shared professional decisions

>* Consistent formatting supports repeated engineering outputs
>* Labels, digits, spacing, and units prevent errors



```
#@title Python Code - Formatted Results

# Format results for civil engineering reports.
# Labels and units prevent output confusion.
# Rounded values support clear professional communication.

# Store computed values from simple engineering calculations.
beam_deflection_mm = 12.487392817
allowable_deflection_mm = 15.0
safety_margin_mm = allowable_deflection_mm - beam_deflection_mm

# Store a second result using Imperial project units.
road_length_ft = 2640.0
road_length_miles = road_length_ft / 5280.0
asphalt_cost = 184735.628

# Format numbers with useful precision and separators.
deflection_text = f"Beam deflection: {beam_deflection_mm:.2f} mm"
margin_text = f"Remaining deflection allowance: {safety_margin_mm:.2f} mm"
length_text = f"Road section length: {road_length_miles:.2f} miles"

# Format currency for report-style cost communication.
cost_text = f"Estimated asphalt cost: ${asphalt_cost:,.2f}"
status_text = f"Design check passed: {beam_deflection_mm <= allowable_deflection_mm}"

# Print a concise engineering report summary.
print("FORMATTED ENGINEERING RESULTS")
print(deflection_text)
print(margin_text)
print(length_text)
print(cost_text)
print(status_text)

```

### **3.2. Decimal Place Control**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780889616" width="250">



>* Show only meaningful decimal precision
>* Match decimals to data accuracy

>* Round outputs to practical precision
>* Use consistent decimals for easier comparison

>* Choose decimals based on context and safety
>* Round final displays, not intermediate calculations



```
#@title Python Code - Decimal Place Control

# Decimal places improve engineering report clarity.
# Rounded display should match measurement precision.
# Keep full values for later calculations.

# Store project information as readable strings.
project_name = "Campus walkway slab estimate"
result_label = "Concrete volume"
unit_label = "cubic meters"

# Use measurements with realistic metric units.
slab_length_m = 18.75
slab_width_m = 2.40
slab_depth_m = 0.125

# Compute the full precision engineering result.
raw_volume_m3 = slab_length_m * slab_width_m * slab_depth_m
rounded_volume_m3 = round(raw_volume_m3, 2)

# Format values for report style output.
report_line = f"{result_label}: {raw_volume_m3:.2f} {unit_label}"
check_line = f"Rounded storage example: {rounded_volume_m3:.2f} {unit_label}"

# Print concise report-ready engineering lines.
print(f"Project: {project_name}")
print(f"Raw calculated volume: {raw_volume_m3} {unit_label}")
print(report_line)
print(check_line)
print("Use formatting for display; keep raw values for calculations.")

```
