## **1. Engineering Variables**

### **1.1. Meaningful Variable Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1780887917" width="250">



>* Use names that explain engineering quantities
>* Clear names improve review and quality

>* Clear names separate similar-looking engineering values
>* Good naming reduces confusion and hidden errors

>* Use clear, consistent, context-rich variable names
>* Treat names as built-in calculation documentation



```
#@title Python Code - Meaningful Variable Names

# Meaningful names make engineering calculations clearer.
# Variable names should describe physical quantities.
# Units in names reduce calculation confusion.

# Define beam dimensions using metric units.
beam_length_m = 6.0
beam_width_m = 0.30
beam_depth_m = 0.50

# Define material density with clear units.
concrete_density_kg_per_m3 = 2400
beam_volume_m3 = beam_length_m * beam_width_m * beam_depth_m
beam_mass_kg = beam_volume_m3 * concrete_density_kg_per_m3

# Define pavement quantities using imperial units.
road_length_ft = 1200
road_width_ft = 24
asphalt_thickness_in = 4

# Convert thickness before computing volume.
asphalt_thickness_ft = asphalt_thickness_in / 12
pavement_volume_ft3 = road_length_ft * road_width_ft * asphalt_thickness_ft

# Print concise engineering results.
print("Beam volume (m^3):", round(beam_volume_m3, 2))
print("Beam mass (kg):", round(beam_mass_kg, 1))
print("Pavement volume (ft^3):", round(pavement_volume_ft3, 1))
print("Names show quantity, context, and units.")

```

### **1.2. Meaningful Variable Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1780887920" width="250">



>* Clear names explain engineering quantities
>* Good naming prevents future confusion

>* Name quantities to prevent engineering mix-ups
>* Show object, property, and calculation relationships

>* Use standard names for team clarity
>* Keep names concise, contextual, and unambiguous



```
#@title Python Code - Meaningful Variable Names

# Meaningful names make engineering calculations easier.
# Stored values should describe real quantities.
# Clear labels reduce review and design errors.

# Define dimensions using descriptive variable names.
beam_span_meters = 6.0
slab_width_meters = 3.5
slab_thickness_meters = 0.18

# Define material density with units in the name.
concrete_unit_weight_kN_per_m3 = 24.0
slab_volume_cubic_meters = slab_width_meters * beam_span_meters * slab_thickness_meters
slab_weight_kN = slab_volume_cubic_meters * concrete_unit_weight_kN_per_m3

# Compare clear names with vague alternatives.
a = 6.0
b = 3.5
c = 0.18

# Compute the same result using unclear names.
d = a * b * c * concrete_unit_weight_kN_per_m3
clear_result = round(slab_weight_kN, 2)
vague_result = round(d, 2)

# Print concise results for easy review.
print("Clear variable names describe the engineering quantities.")
print("Slab volume:", round(slab_volume_cubic_meters, 3), "m^3")
print("Slab weight:", clear_result, "kN")
print("Vague calculation gives:", vague_result, "kN")
print("Prefer names like slab_thickness_meters over c.")

```

### **1.3. Meaningful Variable Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1780887918" width="250">



>* Names link code to real quantities
>* Clear labels make calculations easier to verify

>* Name the object and measured property

>* Clear names support review and collaboration
>* Use consistent names and visible units



```
#@title Python Code - Meaningful Variable Names

# Clear names make calculations easier.
# Engineering quantities should identify meaning.
# Units can appear inside names.

# Store civil engineering quantities with descriptive names.
beam_span_meters = 6.0
concrete_strength_mpa = 30.0
uniform_load_kilonewtons_per_meter = 12.0

# Use the names inside a simple beam expression.
maximum_moment_kilonewton_meters = (
    uniform_load_kilonewtons_per_meter * beam_span_meters ** 2 / 8
)

# Compare with vague names doing the same calculation.
x = 6.0
y = 12.0
z = y * x ** 2 / 8

# Print the meaningful engineering calculation results.
print("Beam span (m):", beam_span_meters)
print("Concrete strength (MPa):", concrete_strength_mpa)
print("Uniform load (kN/m):", uniform_load_kilonewtons_per_meter)

# Show why the descriptive result is easier.
print("Maximum moment (kN*m):", maximum_moment_kilonewton_meters)
print("Same value from vague names:", z)

# Finish with a practical naming reminder.
print("Prefer names that state quantity, location, and units.")

```

## **2. Engineering Expressions**

### **2.1. Arithmetic Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780887925" width="250">



>* Turn measurements into engineering values
>* Write repeatable, checkable arithmetic expressions

>* Follow operation order to avoid wrong results
>* Use parentheses to clarify engineering calculations

>* Match calculations to real engineering meaning
>* Check units before combining quantities



```
#@title Python Code - Arithmetic Calculations

# Arithmetic expressions turn measurements into engineering quantities.
# Parentheses make calculation intent clear and reviewable.
# Unit consistency prevents many common engineering mistakes.

# Store road dimensions using meaningful variable names.
road_length_m = 125.0
road_width_m = 7.5
asphalt_thickness_cm = 12.0

# Convert centimeters to meters before volume calculation.
asphalt_thickness_m = asphalt_thickness_cm / 100.0
pavement_area_m2 = road_length_m * road_width_m
asphalt_volume_m3 = pavement_area_m2 * asphalt_thickness_m

# Combine unit price and fixed cost using parentheses.
unit_price_per_m3 = 145.0
mobilization_cost = 2500.0
total_cost = (asphalt_volume_m3 * unit_price_per_m3) + mobilization_cost

# Use built-in round for readable engineering results.
rounded_area = round(pavement_area_m2, 2)
rounded_volume = round(asphalt_volume_m3, 2)
rounded_cost = round(total_cost, 2)

# Compare division and integer division carefully.
crew_productivity_m3_per_day = 32.0
estimated_days = asphalt_volume_m3 / crew_productivity_m3_per_day
full_days_needed = int(estimated_days) + 1

# Print concise results for quick checking.
print("Pavement area (m^2):", rounded_area)
print("Asphalt volume (m^3):", rounded_volume)
print("Estimated total cost ($):", rounded_cost)
print("Exact crew duration (days):", round(estimated_days, 2))
print("Planned whole days needed:", full_days_needed)

```

### **2.2. Built In Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780887924" width="250">



>* Python functions simplify common engineering calculations
>* They make expressions clearer and more reliable

>* Summarize engineering data with built-in functions
>* Round values and measure differences clearly

>* Use functions with engineering judgment
>* Check inputs, outputs, and rounding timing



```
#@title Python Code - Built In Functions

# Built-in functions simplify civil engineering calculations.
# Small measurement lists support quick engineering summaries.
# Type checks help avoid calculation mistakes.

# Store concrete strengths in megapascals.
strengths_mpa = [31.8, 34.2, 29.9, 36.1]

# Store slab thickness readings in inches.
thickness_inches = [5.9, 6.1, 5.8, 6.0]

# Validate both datasets before summarizing.
data_ready = len(strengths_mpa) > 0 and len(thickness_inches) > 0

# Use built-ins to summarize design information.
max_strength = max(strengths_mpa)
min_thickness = min(thickness_inches)
reading_count = len(thickness_inches)

# Compute averages using sum and len.
average_strength = sum(strengths_mpa) / len(strengths_mpa)
average_thickness = sum(thickness_inches) / len(thickness_inches)

# Use abs to check deviation magnitude.
target_thickness = 6.0
deviation_inches = abs(average_thickness - target_thickness)

# Round results for readable reporting.
reported_strength = round(average_strength, 1)
reported_deviation = round(deviation_inches, 2)

# Check result types before reporting values.
strength_type = type(reported_strength).__name__
count_type = type(reading_count).__name__

print("Data ready:", data_ready)
print("Maximum strength (MPa):", max_strength)
print("Average strength (MPa):", reported_strength)
print("Minimum thickness (in):", min_thickness)
print("Thickness reading count:", reading_count)
print("Average deviation from target (in):", reported_deviation)
print("Reported strength type:", strength_type)
print("Reading count type:", count_type)

```

### **2.3. Using Builtin Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780887922" width="250">



>* Use functions for common engineering tasks
>* Make calculations clearer and less error-prone

>* Match functions to inputs and outputs
>* Round results to practical precision

>* Use functions with engineering judgment.
>* Check timing, units, context, and meaning.



```
#@title Python Code - Using Builtin Functions

# Builtin functions simplify engineering calculations.
# Meaningful variables keep units clear.
# Final rounding supports readable reports.

# Store beam deflections measured in millimeters.
deflection_mm_values = [8.4, 12.7, 10.9, 9.8]

# Store excavation depths measured in feet.
depth_ft_values = [6.2, 5.7, 6.8, 6.1]

# Use max to find the governing deflection.
maximum_deflection_mm = max(deflection_mm_values)

# Use min to find the shallowest excavation point.
minimum_depth_ft = min(depth_ft_values)

# Compare observed and design elevations in meters.
observed_elevation_m = 102.36
design_elevation_m = 102.50

# Use abs to report deviation magnitude only.
elevation_difference_m = abs(observed_elevation_m - design_elevation_m)

# Compute concrete volume from metric dimensions.
slab_length_m = 12.0
slab_width_m = 4.5
slab_thickness_m = 0.18

# Round final volume for practical reporting.
concrete_volume_m3 = slab_length_m * slab_width_m * slab_thickness_m
reported_volume_m3 = round(concrete_volume_m3, 2)

# Print concise engineering results.
print("Maximum deflection:", maximum_deflection_mm, "mm")
print("Minimum excavation depth:", minimum_depth_ft, "ft")
print("Elevation deviation:", round(elevation_difference_m, 3), "m")
print("Reported concrete volume:", reported_volume_m3, "m^3")

```

## **3. Built In Functions**

### **3.1. Rounding Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780887915" width="250">



>* Round values only for clear presentation
>* Keep precision during intermediate calculations

>* Round results to needed decimal places
>* Floating-point approximations can affect boundary rounding

>* Match rounding to the engineering purpose
>* Treat counts differently from measured values



```
#@title Python Code - Rounding Results

# Rounding should support clear engineering communication.
# Keep precise values during intermediate calculations.
# Round only when reporting final results.

# Store a computed concrete volume.
slab_length_m, slab_width_m, slab_depth_m = 8.0, 3.5, 0.127

# Calculate the precise volume first.
volume_m3 = slab_length_m * slab_width_m * slab_depth_m

# Round only for report presentation.
reported_volume_m3 = round(volume_m3, 2)

# Compare precise and rounded values.
print("Precise slab volume:", volume_m3, "m^3")
print("Reported slab volume:", reported_volume_m3, "m^3")

# Compute a required whole truck count.
truck_capacity_m3, safety_factor = 5.0, 1.10

# Use division before choosing a count.
needed_trucks_float = volume_m3 * safety_factor / truck_capacity_m3

# Import a built-in module for upward rounding.
import math

# Round upward because partial trucks are impossible.
needed_trucks_count = math.ceil(needed_trucks_float)

# Show type differences and engineering meaning.
print("Truck calculation value:", round(needed_trucks_float, 3))
print("Truck count to order:", needed_trucks_count)

# Show the resulting Python types.
print("Rounded volume type:", type(reported_volume_m3).__name__)
print("Truck count type:", type(needed_trucks_count).__name__)

```

### **3.2. Absolute Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780887911" width="250">



>* Absolute value shows magnitude, not direction.
>* Use it for tolerances and error checks.

>* Keep number types clear after absolute value
>* Preserve signs when direction still matters

>* Use magnitudes for threshold checks
>* Keep sign meaning before removing it



```
#@title Python Code - Absolute Values

# Absolute value keeps engineering magnitudes clear.
# Signs still carry useful direction information.
# Types affect how calculations are interpreted.

# Store a signed beam deflection measurement.
measured_deflection_mm = -7.5
allowable_deflection_mm = 6

# Keep signed value before removing direction.
deflection_magnitude_mm = abs(measured_deflection_mm)
within_limit = deflection_magnitude_mm <= allowable_deflection_mm

# Compare surveyed and design elevations safely.
design_elevation_m = 102.35
surveyed_elevation_m = 102.18

elevation_error_m = surveyed_elevation_m - design_elevation_m
absolute_error_m = abs(elevation_error_m)

# Show values and their basic Python types.
print("Signed deflection:", measured_deflection_mm, type(measured_deflection_mm).__name__)
print("Deflection magnitude:", deflection_magnitude_mm, type(deflection_magnitude_mm).__name__)
print("Allowable limit:", allowable_deflection_mm, type(allowable_deflection_mm).__name__)
print("Within deflection limit:", within_limit)

# Show direction before showing magnitude.
print("Signed elevation error:", round(elevation_error_m, 3), "m")
print("Absolute elevation error:", round(absolute_error_m, 3), "m")

# Compute a whole-number cover deviation example.
specified_cover_mm = 50
measured_cover_mm = 46

cover_difference_mm = measured_cover_mm - specified_cover_mm
cover_magnitude_mm = abs(cover_difference_mm)

# Print final compliance-style magnitude check.
print("Concrete cover deviation magnitude:", cover_magnitude_mm, "mm")

```

### **3.3. Type Safe Calculations**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780887913" width="250">



>* Check each value’s type before calculating
>* Convert text data to prevent errors

>* Convert outside data before engineering calculations
>* Match data types to quantity meaning

>* Check data types to prevent hidden errors
>* Clean inputs for reliable engineering calculations



```
#@title Python Code - Type Safe Calculations

# Type checks protect engineering calculations.
# Text readings often need conversion.
# Builtins reveal and repair data types.

# Store imported measurements as text strings.
span_text = "24.5"
width_text = "12.0"
depth_text = "0.75"

# Check one original value type.
print("span_text type:", type(span_text).__name__)

# Convert measured quantities to decimal numbers.
span_ft = float(span_text)
width_in = float(width_text)
depth_in = float(depth_text)

# Convert inches to feet before multiplying.
width_ft = width_in / 12.0
depth_ft = depth_in / 12.0
volume_cubic_feet = span_ft * width_ft * depth_ft

# Convert a count string to integer.
bolt_count_text = "8"
bolt_count = int(bolt_count_text)

# Use round for readable engineering output.
rounded_volume = round(volume_cubic_feet, 3)
print("span_ft type:", type(span_ft).__name__)
print("bolt_count type:", type(bolt_count).__name__)
print("concrete volume:", rounded_volume, "ft^3")
print("anchor bolts:", bolt_count)

```
