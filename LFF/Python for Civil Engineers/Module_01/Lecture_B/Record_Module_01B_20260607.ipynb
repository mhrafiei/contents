## **1. Engineering Variables**

### **1.1. Meaningful Variable Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_01.jpg?v=1780879917" width="250">



>* Use names that identify engineering quantities
>* Clear names make calculations easier to review

>* Name quantities with useful context
>* Keep names specific, not overly long

>* Define quantities clearly before using them
>* Prevent confusion and improve engineering documentation



```
#@title Python Code - Meaningful Variable Names

# Meaningful variable names improve engineering calculation clarity.
# Names should describe quantities, units, and context.
# This example compares vague and clear naming.

# Define a beam span using a clear name.
beam_span_length_meters = 6.0

# Define loads with purpose and units visible.
dead_load_kilonewtons_per_meter = 12.5
live_load_kilonewtons_per_meter = 8.0

# Combine named quantities into a readable expression.
total_service_load_kilonewtons_per_meter = (
    dead_load_kilonewtons_per_meter +
    live_load_kilonewtons_per_meter
)

# Compute simply supported beam maximum moment.
maximum_moment_kilonewton_meters = (
    total_service_load_kilonewtons_per_meter *
    beam_span_length_meters ** 2 / 8
)

# Print values with names that explain meaning.
print("Beam span length (m):", beam_span_length_meters)
print("Dead load (kN/m):", dead_load_kilonewtons_per_meter)
print("Live load (kN/m):", live_load_kilonewtons_per_meter)

# Print the final engineering result clearly.
print("Total service load (kN/m):", total_service_load_kilonewtons_per_meter)
print("Maximum moment (kN-m):", maximum_moment_kilonewton_meters)

```

### **1.2. Meaningful Variable Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_02.jpg?v=1780879909" width="250">



>* Name variables by their engineering quantity
>* Clear names prevent confusion during review

>* Name the quantity and engineering context
>* Clear names make calculations readable

>* Use consistent, clear, manageable variable names
>* Names document assumptions and support teamwork



```
#@title Python Code - Meaningful Variable Names

# This example shows meaningful engineering variable names.
# Clear names make calculations easier to review.
# Units in names help prevent mistakes.

# Store bridge dimensions using descriptive names.
bridge_span_length_m = 30.0
bridge_deck_width_m = 12.0
pavement_thickness_m = 0.18

# Store material property with context and units.
concrete_unit_weight_kN_per_m3 = 24.0
slab_volume_m3 = bridge_span_length_m * bridge_deck_width_m * pavement_thickness_m
slab_weight_kN = slab_volume_m3 * concrete_unit_weight_kN_per_m3

# Compare unclear and meaningful variable names.
size = bridge_span_length_m
bridge_span_length_for_review_m = bridge_span_length_m

# Report results using readable engineering labels.
print("Meaningful variable name example")
print(f"Bridge span length: {bridge_span_length_m} m")
print(f"Bridge deck width: {bridge_deck_width_m} m")
print(f"Pavement thickness: {pavement_thickness_m} m")

# Show computed values with units.
print(f"Slab volume: {slab_volume_m3:.2f} m^3")
print(f"Slab weight: {slab_weight_kN:.1f} kN")
print(f"Vague name size: {size} m")
print(f"Clear name span: {bridge_span_length_for_review_m} m")

```

### **1.3. Descriptive Quantity Names**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_01_03.jpg?v=1780879919" width="250">



>* Clear names show each engineering quantity
>* Descriptive names reduce confusion in reviews

>* Name both concept and calculation role
>* Prevent mixing similar engineering quantities

>* Names clarify units, assumptions, and context
>* Clear quantities improve review and accuracy



```
#@title Python Code - Descriptive Quantity Names

# Descriptive names clarify engineering calculation quantities.
# Variables should describe context and units.
# Clear names reduce review and revision errors.

# Store bridge deck quantities using descriptive names.
bridge_deck_width_m = 12.0
span_length_m = 30.0
concrete_thickness_m = 0.25

# Compute area and volume from named quantities.
deck_plan_area_m2 = bridge_deck_width_m * span_length_m
concrete_volume_m3 = deck_plan_area_m2 * concrete_thickness_m

# Store loading quantities with clear engineering meaning.
service_dead_load_kN = 850.0
service_live_load_kN = 420.0
load_factor_dead = 1.2

# Compute factored design load using readable names.
load_factor_live = 1.6
factored_dead_load_kN = load_factor_dead * service_dead_load_kN
factored_live_load_kN = load_factor_live * service_live_load_kN

# Combine quantities only after meanings are clear.
total_factored_load_kN = factored_dead_load_kN + factored_live_load_kN
average_factored_pressure_kPa = total_factored_load_kN / deck_plan_area_m2

# Print a compact engineering summary.
print("Bridge deck area:", round(deck_plan_area_m2, 2), "m^2")
print("Concrete volume:", round(concrete_volume_m3, 2), "m^3")
print("Total factored load:", round(total_factored_load_kN, 2), "kN")
print("Average factored pressure:", round(average_factored_pressure_kPa, 2), "kPa")

```

## **2. Engineering Expressions**

### **2.1. Arithmetic Expressions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_01.jpg?v=1780879921" width="250">



>* Combine quantities to model engineering relationships
>* Use expressions for repeatable engineering calculations

>* Use parentheses to show calculation order
>* Clear expressions improve accuracy and communication

>* Check results for realistic engineering meaning
>* Read each expression as a design statement



```
#@title Python Code - Arithmetic Expressions

# Arithmetic expressions compute engineering quantities.
# Parentheses make calculations easier to read.
# Units should be checked after computing.

# Store beam load components clearly.
dead_load_kN = 42.0
live_load_kN = 28.0
beam_length_m = 7.0

# Add loads before dividing by length.
total_load_kN = dead_load_kN + live_load_kN
load_per_meter_kN_m = total_load_kN / beam_length_m

# Compute slab volume using multiplication.
slab_length_ft = 30.0
slab_width_ft = 18.0
slab_thickness_ft = 0.5

# Multiply dimensions to get cubic feet.
slab_volume_ft3 = slab_length_ft * slab_width_ft * slab_thickness_ft
slab_volume_yd3 = slab_volume_ft3 / 27.0

# Use exponents for circular footing area.
footing_radius_m = 1.2
footing_area_m2 = 3.14159 * footing_radius_m ** 2

# Compute pressure using grouped total load.
column_load_kN = 950.0
bearing_pressure_kPa = column_load_kN / footing_area_m2

# Print concise engineering results.
print("Beam total load:", round(total_load_kN, 1), "kN")
print("Beam line load:", round(load_per_meter_kN_m, 1), "kN/m")
print("Concrete volume:", round(slab_volume_yd3, 2), "yd^3")
print("Footing area:", round(footing_area_m2, 2), "m^2")
print("Bearing pressure:", round(bearing_pressure_kPa, 1), "kPa")

```

### **2.2. Built In Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_02.jpg?v=1780879923" width="250">



>* Use functions for common engineering calculations
>* Round, compare, and summarize values clearly

>* Report results with suitable precision
>* Use functions to clarify engineering decisions

>* Match functions to appropriate engineering inputs
>* Check units, reasonableness, and physical meaning



```
#@title Python Code - Built In Functions

# Built-in functions support engineering expression checks.
# Civil engineers often summarize calculated quantities.
# Meaningful variables keep calculations readable.

# Store footing dimensions using metric units.
footing_length_m = 3.6
footing_width_m = 2.4
footing_depth_m = 0.55

# Compute concrete volume using arithmetic expression.
concrete_volume_m3 = footing_length_m * footing_width_m * footing_depth_m
rounded_volume_m3 = round(concrete_volume_m3, 2)

# Compare trial beam reactions using built-in functions.
trial_reactions_kN = [84.2, 91.7, 88.5, 93.1]
governing_reaction_kN = max(trial_reactions_kN)
smallest_reaction_kN = min(trial_reactions_kN)

# Use absolute value for unsigned difference.
allowable_settlement_mm = 25.0
predicted_settlement_mm = 31.4
settlement_excess_mm = abs(predicted_settlement_mm - allowable_settlement_mm)

# Count inspections before computing average safely.
inspection_deflections_mm = [4.8, 5.1, 4.9]
reading_count = len(inspection_deflections_mm)
average_deflection_mm = sum(inspection_deflections_mm) / reading_count

# Display concise engineering results.
print("Rounded concrete volume:", rounded_volume_m3, "m^3")
print("Governing reaction:", governing_reaction_kN, "kN")
print("Smallest reaction:", smallest_reaction_kN, "kN")
print("Settlement difference magnitude:", settlement_excess_mm, "mm")
print("Average measured deflection:", round(average_deflection_mm, 2), "mm")

```

### **2.3. Unit Conversions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_02_03.jpg?v=1780879924" width="250">



>* Convert units within engineering calculations.
>* Prevent major errors from unit mistakes.

>* Match conversion method to measurement type
>* Use expressions and functions for accuracy

>* Check converted values for physical reasonableness
>* Use clear expressions to avoid unit errors



```
#@title Python Code - Unit Conversions

# Unit conversions support engineering expressions.
# Variables should name quantities and units.
# Built-ins help report results clearly.

# Store field measurements using clear variable names.
road_length_ft = 5280
beam_depth_in = 18
concrete_volume_yd3 = 12

# Define conversion factors as named variables.
feet_to_meters = 0.3048
inch_to_mm = 25.4
yd3_to_m3 = 0.764555

# Convert each quantity with arithmetic expressions.
road_length_m = road_length_ft * feet_to_meters
beam_depth_mm = beam_depth_in * inch_to_mm
concrete_volume_m3 = concrete_volume_yd3 * yd3_to_m3

# Use built-ins for readable engineering reporting.
rounded_road_m = round(road_length_m, 1)
rounded_volume_m3 = round(concrete_volume_m3, 2)
largest_metric_value = max(road_length_m, beam_depth_mm, concrete_volume_m3)

# Print compact results for quick checking.
print("Road length:", rounded_road_m, "m")
print("Beam depth:", beam_depth_mm, "mm")
print("Concrete volume:", rounded_volume_m3, "m^3")
print("Largest numeric result:", round(largest_metric_value, 1))
print("Road length type:", type(road_length_m).__name__)

```

## **3. Built In Functions**

### **3.1. Rounding Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_01.jpg?v=1780879907" width="250">



>* Round results to match practical reporting precision
>* Floating-point values may need controlled display

>* Separate display rounding from calculation rounding
>* Keep precision until final engineering results

>* Rounded values may still be floats
>* Choose rounding based on real-world counts



```
#@title Python Code - Rounding Results

# Rounding supports clear engineering communication.
# Float results often need careful handling.
# Keep precision until final reporting.

# Store slab dimensions as metric variables.
slab_length_m = 12.4
slab_width_m = 7.8
slab_thickness_m = 0.18

# Compute volume using full float precision.
volume_m3 = slab_length_m * slab_width_m * slab_thickness_m
rounded_volume_m3 = round(volume_m3, 2)

# Compare display rounding and reused rounding.
bays = 20
full_total_m3 = volume_m3 * bays
rounded_total_m3 = rounded_volume_m3 * bays

# Show float type and readable rounded output.
print("Raw volume:", volume_m3, type(volume_m3))
print("Rounded volume:", rounded_volume_m3, type(rounded_volume_m3))
print("Full total:", round(full_total_m3, 2), "m^3")

# Show effect of rounding before multiplication.
print("Rounded-first total:", round(rounded_total_m3, 2), "m^3")
print("Difference:", round(full_total_m3 - rounded_total_m3, 4), "m^3")

# Show a count situation using built-in round.
truck_capacity_m3 = 6.0
truckloads_float = full_total_m3 / truck_capacity_m3
print("Truckloads rounded:", round(truckloads_float), "may underestimate.")

```

### **3.2. Absolute Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_02.jpg?v=1780879905" width="250">



>* Absolute value shows magnitude, not sign
>* Use it when size matters most

>* Absolute value keeps the input’s numeric type
>* It changes sign, not precision or meaning

>* Use magnitudes for tolerances and errors
>* Keep signs when direction matters



```
#@title Python Code - Absolute Values

# Absolute value measures magnitude without direction.
# Civil quantities often keep meaningful signs.
# Type awareness prevents hidden calculation mistakes.

# Store signed beam deflection errors.
predicted_deflection_in = 0.42
measured_deflection_in = -0.48
signed_error_in = measured_deflection_in - predicted_deflection_in

# Use abs for magnitude only.
error_magnitude_in = abs(signed_error_in)
print("Beam signed error, inches:", signed_error_in)
print("Beam absolute error, inches:", error_magnitude_in)

# Compare measured elevation differences.
design_elevation_m = 102.75
survey_elevation_m = 102.68
elevation_difference_m = survey_elevation_m - design_elevation_m

# Preserve decimal precision after abs.
elevation_magnitude_m = abs(elevation_difference_m)
print("Elevation magnitude, meters:", elevation_magnitude_m)
print("Elevation result type:", type(elevation_magnitude_m).__name__)

# Compare ordered and delivered counts.
ordered_bolts = 120
delivered_bolts = 116
bolt_count_difference = delivered_bolts - ordered_bolts

# Integer inputs keep integer magnitude.
missing_bolt_magnitude = abs(bolt_count_difference)
print("Bolt count magnitude:", missing_bolt_magnitude)
print("Bolt result type:", type(missing_bolt_magnitude).__name__)

# Check a construction tolerance.
thickness_error_cm = -0.35
allowed_tolerance_cm = 0.50
within_tolerance = abs(thickness_error_cm) <= allowed_tolerance_cm

# Print the tolerance decision.
print("Thickness within tolerance:", within_tolerance)

```

### **3.3. Type Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Civil Engineers/Module_01/Lecture_B/image_03_03.jpg?v=1780879904" width="250">



>* Check value types before engineering calculations
>* Prevent text-number mix-ups and misleading results

>* Separate calculation numbers from descriptive text
>* Check imported values before engineering computations

>* Distinguish whole-number counts from decimal measurements
>* Check types to debug and validate calculations



```
#@title Python Code - Type Checks

# Type checks protect engineering calculations.
# Values can look deceptively similar.
# Python types guide safe operations.

# Store example civil engineering values.
span_m = 24.5
pile_count = 8
beam_label = "B12"

# Check each value's Python type.
print("span_m type:", type(span_m).__name__)
print("pile_count type:", type(pile_count).__name__)
print("beam_label type:", type(beam_label).__name__)

# Simulate a measurement imported as text.
thickness_cm_text = "18.5"
is_numeric_before = isinstance(thickness_cm_text, (int, float))
print("numeric before conversion:", is_numeric_before)

# Convert text before arithmetic calculation.
thickness_cm = float(thickness_cm_text)
area_m2 = 42.0
volume_m3 = area_m2 * thickness_cm / 100

# Confirm converted value supports arithmetic.
is_numeric_after = isinstance(thickness_cm, (int, float))
print("numeric after conversion:", is_numeric_after)
print("concrete volume m3:", round(volume_m3, 2))

# Check whole-number quantities separately.
valid_pile_count = isinstance(pile_count, int)
print("pile count is integer:", valid_pile_count)
print("ready for calculation:", is_numeric_after and valid_pile_count)

```
