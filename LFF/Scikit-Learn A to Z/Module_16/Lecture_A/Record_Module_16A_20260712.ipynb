## **2. Split Criteria**

### **2.1. Choosing Split Criteria**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_16/Lecture_A/image_02_01.jpg?v=1783848693" width="250">



>* Criteria judge split quality in trees.
>* They shape accuracy, thresholds, and patterns.

>* Criteria should match data and goals.
>* Compare criteria using validation, not defaults.

>* Split criteria shape rules and interpretation.
>* Prefer stable, sensible, explainable splits over accuracy.



### **2.2. Split Criteria Basics**

>* Split criteria choose the best tree question.
>* Classification seeks purity; regression seeks similarity.

>* Criteria score how much splits improve data.
>* Different criteria change tree structure and performance.

>* Greedy early splits strongly shape tree structure.
>* Split criteria guide rules and tuning choices.



### **2.3. Choosing Split Criteria**

## **3. Interpreting Decision Trees**

### **3.1. Tree Visualization**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_16/Lecture_A/image_03_01.jpg?v=1783848689" width="250">



>* Visual trees show prediction paths clearly.
>* Useful for explaining decisions in practice.

>* Nodes show counts, purity, and predictions.
>* Colors and sample sizes reveal trustworthiness.

>* Read trees critically for complexity and overfitting.
>* Top splits reveal influential features and patterns.



### **3.2. Reading Tree Rules**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_16/Lecture_A/image_03_02.jpg?v=1783848694" width="250">



>* Paths become step-by-step if-then decisions.
>* Rules show predictions and triggering conditions.

>* Rules depend on split order and thresholds
>* Interpret full path within its context

>* Judge rules for reliability, fairness, and limits.
>* Useful rules support explanation and validation.



### **3.3. Feature Split Analysis**
