### **1.2. Choosing SGD Estimators**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_01_02.jpg?v=1783843926" width="250">



>* Choose classifier or regressor by target type.
>* SGD suits large, fast, memory-efficient learning.

>* Best for sparse, high-dimensional, regularly updated data
>* Less suitable for strongly nonlinear relationships

>* Choose SGD for fast, scalable incremental learning.

### **1.3. Choosing Loss Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_01_03.jpg?v=1783843927" width="250">



>* Loss turns goals into training signals.
>* Choose classification or regression loss carefully.

>* Logistic favors probabilities; hinge emphasizes class margins.
>* Choose based on multiclass needs and priorities.

>* Choose regression loss based on error costs.
>* Squared punishes outliers; absolute is more robust.



## **2. Optimization Controls**

### **2.1. Learning Rate Tuning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_02_01.jpg?v=1783843916" width="250">



>* Learning rate controls update step size.
>* Balance speed, stability, and generalization.

>* Watch loss and validation for stability.
>* Balance adaptation speed against noisy overreaction.

>* Decrease learning rate as training progresses.
>* Tune with validation and training context.



### **2.2. Learning Rate Tuning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_02_02.jpg?v=1783843912" width="250">



>* Learning rate controls update step size.
>* Too high destabilizes; too low stalls.

>* Learning rate affects generalization, not just speed.
>* Tune using validation stability and consistency.

>* Search learning rates; schedules often work best.
>* Aim for stable convergence and strong generalization.



### **2.3. Early Stopping**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_02_03.jpg?v=1783843914" width="250">



>* Stop when validation performance stops improving.
>* Prevents overfitting and preserves generalization.

>* Too much training memorizes noise, not patterns.
>* Stop at validation plateau to preserve generalization.

>* Use patience; validation scores often fluctuate.
>* Learning rate, penalties affect stopping and stability.



## **3. Incremental Learning**

### **3.1. Partial Fit Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_03_01.jpg?v=1783843918" width="250">



>* Learns in small batches, not all data.
>* Updates existing knowledge as new labels arrive.

>* Partial fit updates models batch by batch.
>* Batch order and known classes matter.

>* Monitor batches carefully to avoid drift.
>* Partial fit enables fast, continuous adaptation.



### **3.2. Streaming Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_03_02.jpg?v=1783843920" width="250">



>* partial_fit updates models as new data arrives
>* Useful for changing, high-volume prediction tasks

>* Keep preprocessing consistent across incoming batches.
>* Watch imbalance, ordering, and shifting data.

>* Models learn continuously while staying in service.
>* Monitor drift and balance recency with stability.



### **3.3. Streaming Updates**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_B/image_03_03.jpg?v=1783843922" width="250">



>* partial_fit updates models as data arrives.
>* Useful for large, changing data streams.

>* partial_fit makes small updates from each batch.
>* Batch order matters; monitor drifting performance.

>* Keep features consistent and classes predefined.
>* Monitor streams regularly to adapt models.
