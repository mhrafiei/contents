## **3. Discriminant Analysis**

### **3.1. Linear Boundaries**

>* Linear boundaries separate classes with hyperplanes.
>* Logistic regression and LDA can align.

>* Shared covariance makes LDA boundaries linear.
>* Linear boundaries are simple and interpretable.

>* Linear boundaries trade detail for robustness.
>* Best when classes differ mainly by location.



### **3.2. Quadratic Boundaries**

### **3.3. Covariance Assumptions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_11/Lecture_A/image_03_03.jpg?v=1783843461" width="250">



>* Covariance describes joint feature spread within classes.
>* LDA shares covariance, giving stable linear boundaries.

>* QDA lets each class vary differently.
>* Flexible, but needs more data.

>* Logistic regression skips class covariance modeling.
>* Choose LDA, QDA, or logistic by assumptions.



```
#@title Python Code - Covariance Assumptions

# This script compares covariance assumptions visually.
# It uses tiny synthetic two dimensional classes.
# One plot shows linear and curved boundaries.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Define class means and covariances.
mean_a = np.array([0.0, 0.0])
mean_b = np.array([2.2, 1.2])

# Use one shared covariance for LDA.
shared_cov = np.array([[1.0, 0.6], [0.6, 1.0]])

# Use different covariances for QDA.
qda_cov_a = np.array([[1.0, 0.6], [0.6, 1.0]])
qda_cov_b = np.array([[1.4, -0.5], [-0.5, 0.5]])

# Draw small samples for both classes.
class_a = rng.multivariate_normal(mean_a, qda_cov_a, size=80)
class_b = rng.multivariate_normal(mean_b, qda_cov_b, size=80)

# Check sample shapes before plotting.
assert class_a.shape == (80, 2)
assert class_b.shape == (80, 2)

# Build a small prediction grid.
x1 = np.linspace(-3.5, 5.0, 220)
x2 = np.linspace(-3.5, 4.0, 220)

# Create coordinate matrices safely.
xx, yy = np.meshgrid(x1, x2)
grid = np.c_[xx.ravel(), yy.ravel()]

# Define a stable Gaussian score.
def gaussian_score(x, mean, cov):

    inv_cov = np.linalg.inv(cov)
    sign, logdet = np.linalg.slogdet(cov)

    diff = x - mean
    quad = np.einsum("...i,ij,...j->...", diff, inv_cov, diff)

    return -0.5 * (quad + logdet)

# Compute LDA scores using shared covariance.
lda_a = gaussian_score(grid, mean_a, shared_cov)
lda_b = gaussian_score(grid, mean_b, shared_cov)

# Compute QDA scores using class covariances.
qda_a = gaussian_score(grid, mean_a, qda_cov_a)
qda_b = gaussian_score(grid, mean_b, qda_cov_b)

# Reshape score differences for contours.
lda_diff = (lda_a - lda_b).reshape(xx.shape)
qda_diff = (qda_a - qda_b).reshape(xx.shape)

# Create one beginner friendly figure.
plt.figure(figsize=(8, 6))
plt.scatter(class_a[:, 0], class_a[:, 1], label="Class A", alpha=0.7)
plt.scatter(class_b[:, 0], class_b[:, 1], label="Class B", alpha=0.7)

# Draw the LDA linear boundary.
plt.contour(xx, yy, lda_diff, levels=[0], colors=["black"], linewidths=2)

# Draw the QDA curved boundary.
plt.contour(xx, yy, qda_diff, levels=[0], colors=["crimson"], linewidths=2)

# Add short labels and guidance.
plt.title("Shared covariance gives linear, separate covariance gives curved")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Keep the legend compact.
plt.legend(["Class A", "Class B", "LDA boundary", "QDA boundary"])
plt.tight_layout()
plt.show()

```
