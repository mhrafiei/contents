### **2.2. Selecting Column Groups**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_06/Lecture_B/image_02_02.jpg?v=1783835381" width="250">



>* Group similar columns for suitable preprocessing.
>* Thoughtful grouping preserves meaning and improves learning.

>* Group columns by names, types, or purpose.
>* Choose groups for preprocessing and domain logic.

>* Group columns by meaning to avoid errors.
>* Clear groups improve pipeline reliability and transparency.



```
#@title Python Code - Selecting Column Groups

# Group columns by meaning and data type.
# Build separate paths for each feature group.
# Keep the example tiny and inspectable.

import pandas as pd
from sklearn.compose import ColumnTransformer

# Create a tiny mixed dataset.
df = pd.DataFrame({
    "age": [25, 40, 31],
    "income": [50000, 80000, 62000],
    "city": ["Boston", "Austin", "Boston"],
    "review": ["fast delivery", "great support", "late package"],
    "member": ["yes", "no", "yes"],
})

# Select groups by explicit column names.
numeric_cols = ["age", "income"]
categorical_cols = ["city", "member"]
text_col = "review"

# Build a column-wise preprocessing plan.
ct = ColumnTransformer(
    transformers=[
        ("numbers", "passthrough", numeric_cols),
        ("labels", "passthrough", categorical_cols),
        ("text", "passthrough", text_col),
    ],
    remainder="drop",
    verbose_feature_names_out=True,
)

# Apply the grouped selection.
Xt = ct.fit_transform(df)
feature_names = ct.get_feature_names_out()

# Show the chosen groups clearly.
print("Numeric group:", numeric_cols)
print("Categorical group:", categorical_cols)
print("Text group:", [text_col])
print("Output shape:", Xt.shape)
print("Feature names:", list(feature_names))

```
