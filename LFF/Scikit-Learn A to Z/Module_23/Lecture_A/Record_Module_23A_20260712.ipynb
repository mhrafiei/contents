## **2. Advanced Density Tuning**

### **2.1. Choosing eps values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_02_01.jpg?v=1783856612" width="250">



>* Eps sets how close points must be.
>* Too small splits; too large merges clusters.

>* Scale features before choosing eps.
>* Prefer eps ranges with stable clusters.

>* Eps depends on domain-specific meaning of closeness.
>* Balance detail, noise tolerance, and cluster separation.



### **2.2. Tuning Min Samples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_02_02.jpg?v=1783856616" width="250">



>* min_samples sets neighborhood evidence required
>* Balance noise resistance against missing small clusters

>* min_samples controls local clustering strictness.
>* Low merges; high misses real patterns.

>* Balance errors, domain needs, and data scale.
>* Compare settings for stable, meaningful clusters.



### **2.3. Refining Cluster Size**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_02_03.jpg?v=1783856614" width="250">



>* Cluster size filters meaningful groups from noise.
>* Small values keep detail; large miss niches.

>* Match cluster size to decision scale.
>* Use domain judgment for broad or rare patterns.

>* Cluster size depends on eps and density.
>* Choose settings balancing meaning, stability, and noise.



## **3. Large Scale Clustering**

### **3.1. BIRCH Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_03_01.jpg?v=1783856600" width="250">



>* BIRCH summarizes large datasets efficiently.
>* Builds trees for fast, scalable clustering.

>* Builds compact summaries in a hierarchy.
>* Enables fast large-scale clustering and refinement.

>* BIRCH scales well but misses complex shapes.
>* Useful alone or as preprocessing.



### **3.2. BIRCH Evaluation Tools**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_03_02.jpg?v=1783856607" width="250">



>* Evaluate BIRCH by inspecting CF tree quality.
>* Check subclusters preserve useful distinctions efficiently.

>* Check cluster tightness and separation after summarization.
>* Unstable overlap signals poor BIRCH parameter settings.

>* Measure scalability, memory, speed, and stability.
>* Compare BIRCH quality and efficiency at scale.



### **3.3. Clustering Performance Evaluation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_23/Lecture_A/image_03_03.jpg?v=1783856609" width="250">



>* Speed alone does not ensure quality.
>* Check cluster compactness, separation, and similarity.

>* Combine internal structure with external usefulness.
>* Good clusters support real decisions, not scores.

>* Test clustering stability across scale and changes.
>* Ensure results stay useful and reliable.
