## **2. Elastic Net Stability**

### **2.1. Why Elastic Net**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_09/Lecture_B/image_02_01.jpg?v=1783840698" width="250">



>* Correlated predictors can make Lasso unstable.
>* Elastic Net handles related features better.

>* Lasso can give inconsistent feature stories.
>* Elastic Net shares influence, improving stability.

>* Correlated features often carry shared information.
>* Elastic Net keeps groups while controlling overfitting.



```
#@title Python Code - Why Elastic Net

# Elastic Net handles correlated predictors more smoothly.
# This example compares coefficient stability visually.
# We use tiny synthetic housing style data.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Build correlated features and target.
n = 80
size = rng.normal(1500, 250, n)
rooms = size / 500 + rng.normal(0, 0.15, n)
noise = rng.normal(0, 0.8, n)

# Create a target using both features.
y = 0.04 * size + 6.0 * rooms + noise

# Stack features into one matrix.
X = np.column_stack((size, rooms))

# Standardize features for fair penalties.
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Center the target for simplicity.
y = y - y.mean()

# Check shapes before path calculations.
assert X.shape == (n, 2)
assert y.shape == (n,)

# Import path solvers from sklearn.
from sklearn.linear_model import lasso_path, enet_path

# Compute the Lasso coefficient path.
alphas_lasso, coefs_lasso, _ = lasso_path(X, y)

# Compute the Elastic Net path.
result_enet = enet_path(X, y, l1_ratio=0.6)
alphas_enet, coefs_enet, _ = result_enet

# Create one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the Lasso coefficient path.
axes[0].plot(alphas_lasso, coefs_lasso[0], label="size")
axes[0].plot(alphas_lasso, coefs_lasso[1], label="rooms")
axes[0].set_xscale("log")
axes[0].set_title("Lasso: one feature can dominate")

# Label the first panel clearly.
axes[0].set_xlabel("alpha")
axes[0].set_ylabel("coefficient")
axes[0].legend()
axes[0].invert_xaxis()

# Plot the Elastic Net coefficient path.
axes[1].plot(alphas_enet, coefs_enet[0], label="size")
axes[1].plot(alphas_enet, coefs_enet[1], label="rooms")
axes[1].set_xscale("log")
axes[1].set_title("Elastic Net: shared influence")

# Label the second panel clearly.
axes[1].set_xlabel("alpha")
axes[1].legend()
axes[1].invert_xaxis()

# Add one teaching summary line.
fig.suptitle("Correlated features stay more balanced with Elastic Net")
plt.tight_layout()
plt.show()

```
