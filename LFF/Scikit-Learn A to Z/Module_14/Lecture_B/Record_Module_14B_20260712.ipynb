### **1.3. PLS Covariance Focus**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_14/Lecture_B/image_01_03.jpg?v=1783846948" width="250">



>* PLS finds patterns that vary together.

>* Covariance reflects association strength and variable scale.
>* Scaling choices change which shared patterns dominate.

>* PLS preserves strongest shared variation across datasets.
>* Covariance keeps scale, linking feature sets.



```
#@title Python Code - PLS Covariance Focus

# This script illustrates covariance focused latent directions.
# We compare unscaled and scaled paired features.
# Results show why preprocessing changes interpretation.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create one shared hidden signal.
n_samples = 120
z_shared = rng.normal(size=(n_samples, 1))
noise = rng.normal(scale=0.4, size=(n_samples, 4))

# Build paired feature blocks.
X = np.hstack((8 * z_shared + noise[:, [0]],
               1 * z_shared + noise[:, [1]]))
Y = np.hstack((7 * z_shared + noise[:, [2]],
               1 * z_shared + noise[:, [3]]))

# Center columns before covariance calculations.
Xc = X - X.mean(axis=0, keepdims=True)
Yc = Y - Y.mean(axis=0, keepdims=True)

# Compute cross covariance between blocks.
C_xy = Xc.T @ Yc / (n_samples - 1)
U, S, Vt = np.linalg.svd(C_xy, full_matrices=False)
wx_cov = U[:, 0]
wy_cov = Vt.T[:, 0]

# Form first latent scores.
t_cov = Xc @ wx_cov
u_cov = Yc @ wy_cov
cov_value = np.cov(t_cov, u_cov, ddof=1)[0, 1]
cor_value = np.corrcoef(t_cov, u_cov)[0, 1]

# Standardize each variable for comparison.
Xs = Xc / Xc.std(axis=0, ddof=1, keepdims=True)
Ys = Yc / Yc.std(axis=0, ddof=1, keepdims=True)

# Recompute directions after scaling.
R_xy = Xs.T @ Ys / (n_samples - 1)
Us, Ss, Vts = np.linalg.svd(R_xy, full_matrices=False)
wx_scaled = Us[:, 0]
wy_scaled = Vts.T[:, 0]

# Form scaled latent scores.
t_scaled = Xs @ wx_scaled
u_scaled = Ys @ wy_scaled
cov_scaled = np.cov(t_scaled, u_scaled, ddof=1)[0, 1]
cor_scaled = np.corrcoef(t_scaled, u_scaled)[0, 1]

# Print a short teaching summary.
print("Unscaled X weights:", np.round(wx_cov, 3))
print("Unscaled Y weights:", np.round(wy_cov, 3))
print("Unscaled score covariance:", round(cov_value, 3))
print("Unscaled score correlation:", round(cor_value, 3))
print("Scaled X weights:", np.round(wx_scaled, 3))
print("Scaled Y weights:", np.round(wy_scaled, 3))
print("Scaled score covariance:", round(cov_scaled, 3))
print("Scaled score correlation:", round(cor_scaled, 3))

# Plot latent score relationships.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(t_cov, u_cov, color="tab:blue", alpha=0.8)
axes[0].set_title("PLS style covariance focus")
axes[0].set_xlabel("X latent score")
axes[0].set_ylabel("Y latent score")

# Compare with standardized variables.
axes[1].scatter(t_scaled, u_scaled, color="tab:orange", alpha=0.8)
axes[1].set_title("After variable scaling")
axes[1].set_xlabel("Scaled X score")
axes[1].set_ylabel("Scaled Y score")

# Finish the single teaching plot.
plt.tight_layout()
plt.show()

```

## **3. Component Interpretation**

### **3.1. Understanding Scores**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_14/Lecture_B/image_03_01.jpg?v=1783846924" width="250">



>* Scores locate observations in latent component space.
>* Similar scores indicate similar shared profiles.

>* Scores map observations, revealing clusters and outliers.
>* Scores summarize patterns, not direct measurements.

>* Interpret score signs relative to loadings.
>* Scaling and component count shape meaning.



```
#@title Python Code - Understanding Scores

# Scores show observation positions on latent directions.
# This example builds paired data manually.
# We interpret score signs and clustering.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create one shared hidden pattern.
latent = np.linspace(-2.5, 2.5, 24)
noise_x = rng.normal(0, 0.25, (24, 2))
noise_y = rng.normal(0, 0.25, (24, 2))

# Build two paired feature blocks.
X = np.column_stack((2.0 * latent, -1.2 * latent)) + noise_x
Y = np.column_stack((1.5 * latent, 0.8 * latent)) + noise_y

# Check paired dataset compatibility.
assert X.shape[0] == Y.shape[0] >= 8
assert X.shape[1] == 2 and Y.shape[1] == 2

# Standardize each block separately.
Xz = (X - X.mean(axis=0)) / X.std(axis=0, ddof=1)
Yz = (Y - Y.mean(axis=0)) / Y.std(axis=0, ddof=1)

# Find cross-covariance between blocks.
C = Xz.T @ Yz / (Xz.shape[0] - 1)
U, S, Vt = np.linalg.svd(C, full_matrices=False)

# Compute first-component score coordinates.
t_scores = Xz @ U[:, 0]
u_scores = Yz @ Vt.T[:, 0]
correlation = np.corrcoef(t_scores, u_scores)[0, 1]

# Summarize score interpretation briefly.
print(f"First cross strength: {S[0]:.2f}")
print(f"Score correlation: {correlation:.2f}")
print(f"Highest score case: {np.argmax(t_scores)}")
print(f"Lowest score case: {np.argmin(t_scores)}")
print("Nearby points have similar shared profiles.")
print("Sign can flip without changing meaning.")

# Plot score space for both blocks.
plt.figure(figsize=(6, 4))
plt.scatter(t_scores, u_scores, c=latent, cmap="coolwarm", s=55)
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

# Label axes and title clearly.
plt.xlabel("X block scores")
plt.ylabel("Y block scores")
plt.title("Understanding scores in paired data")
plt.tight_layout()
plt.show()

```
