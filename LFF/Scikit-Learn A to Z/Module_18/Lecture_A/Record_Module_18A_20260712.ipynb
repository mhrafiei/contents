## **3. Histogram Boosting Features**

### **3.1. Native Categorical Support**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_18/Lecture_A/image_03_01.jpg?v=1783850575" width="250">



>* Uses categories directly without one-hot encoding
>* Simplifies preprocessing and keeps models manageable

>* Groups categories instead of encoding or ordering.
>* Captures interactions efficiently in mixed data.

>* Rare or changing categories need caution.
>* Validate fairness, leakage, and category meaning.



### **3.2. Missing Value Handling**

>* Histogram boosting handles missing values directly.
>* Missingness can carry useful information.

>* Each split learns missing values' best branch.
>* Split-specific handling preserves useful missingness patterns.

>* Missing values still require data understanding.
>* Validate changing patterns and systematic gaps.



### **3.3. Missing Value Handling**
