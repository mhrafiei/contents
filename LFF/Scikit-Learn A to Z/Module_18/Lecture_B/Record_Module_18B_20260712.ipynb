### **1.2. Voting Ensemble Basics**

>* Diverse models improve accuracy and robustness.

>* Diverse, capable models make voting useful.
>* Similar or weak models add little.

>* Simple voting combines predictions across tasks.

### **1.3. Voting Strategies**

## **3. Leakage Safe Stacking**

### **3.1. Leakage Free Training**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_18/Lecture_B/image_03_01.jpg?v=1783850891" width="250">



>* Meta-model must avoid same-target predictions.
>* Leakage inflates results and hurts deployment.

>* Use out-of-sample predictions for every training row.
>* This teaches realistic behavior and prevents hindsight.

>* Preprocessing and tuning must avoid leakage.
>* Strict separation makes stacks trustworthy and comparable.



### **3.2. Cross Validated Meta Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_18/Lecture_B/image_03_02.jpg?v=1783850893" width="250">



>* Out-of-fold predictions prevent stacking leakage.
>* Meta models learn realistic unseen performance.

>* Meta features vary by task and model.
>* Out-of-fold predictions preserve realistic model differences.

>* Out-of-fold features enable fair ensemble comparisons.
>* Keep preprocessing folded; retrain base models later.



### **3.3. Leakage Free Evaluation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_18/Lecture_B/image_03_03.jpg?v=1783850895" width="250">



>* Evaluate only on completely untouched data.
>* Use out-of-fold training, then held-out testing.

>* Meta-training cross-validation is not final evaluation.
>* Use nested CV or holdout testing.

>* Compare ensembles using identical, leakage free evaluation.
>* Keep preprocessing and splits realistic across time.
