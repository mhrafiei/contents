## **2. Core Estimator Methods**

### **2.1. Fitting the Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_01.jpg?v=1783902173" width="250">



>* Fit lets estimators learn from training data
>* Models and transformers store learned information

>* Fit only on training data
>* Keep test data unseen for fair evaluation

>* Fitted estimators use learned information later
>* Refitting can replace learned model state



```
#@title Python Code - Fitting the Model

# Fit lets estimators learn from training data.
# This example uses only built-in Python.
# We mimic estimator methods without scikit-learn.

# Define a tiny estimator class.
class MeanRegressor:

    # Store configuration before fitting.
    def __init__(self, unit_name="feet"):
        self.unit_name = unit_name

    # Learn one value from targets.
    def fit(self, X, y):
        if len(X) != len(y):
            raise ValueError("X and y lengths must match")

        if len(y) == 0:
            raise ValueError("Training data cannot be empty")

        self.mean_ = sum(y) / len(y)
        self.n_features_in_ = len(X[0])
        return self

    # Predict using learned information.
    def predict(self, X):
        if not hasattr(self, "mean_"):
            raise RuntimeError("Call fit before predict")

        return [self.mean_ for row in X]

    # Score with simple mean absolute error.
    def score(self, X, y):
        predictions = self.predict(X)
        errors = [abs(a - b) for a, b in zip(y, predictions)]
        return sum(errors) / len(errors)

# Create tiny house training data.
X_train = [[900], [1200], [1500], [1800]]
y_train = [220000, 260000, 310000, 360000]

# Create tiny future-like evaluation data.
X_test = [[1000], [1700]]
y_test = [240000, 340000]

# Configure estimator before fitting.
model = MeanRegressor(unit_name="square feet")
print("Before fit, parameters:", model.unit_name)

# Fit learns attributes ending with underscores.
model.fit(X_train, y_train)
print("After fit, learned mean_:", round(model.mean_))

# Predict and score after fitting.
predictions = model.predict(X_test)
print("Predictions:", [round(value) for value in predictions])

# Lower score means smaller average error here.
mae = model.score(X_test, y_test)
print("Mean absolute error:", round(mae))

# Refit replaces learned information.
model.fit([[2000], [2200]], [390000, 430000])
print("After refit, new mean_:", round(model.mean_))

```

### **2.2. Predict and Score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_02.jpg?v=1783902175" width="250">



>* Predict after fitting to label new data
>* Use matching features from training data

>* Score summarizes fitted model performance
>* Requires known answers for evaluation

>* Predict outputs; score evaluates known results
>* Use unseen data for reliable scoring



```
#@title Python Code - Predict and Score

# Demonstrate predict and score methods.
# Use a tiny fitted estimator.
# Keep outputs short and clear.

import numpy as np

# Define a tiny estimator class.
class TinyThresholdClassifier:

    # Store configurable estimator parameters.
    def __init__(self, threshold=0.0):
        self.threshold = threshold

    # Learn a threshold from labeled data.
    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=int)

        # Validate matching sample counts.
        if X.shape[0] != y.shape[0]:
            raise ValueError("X and y lengths differ")

        # Learn class means as attributes.
        low_mean = X[y == 0].mean()
        high_mean = X[y == 1].mean()

        # Store learned threshold after fitting.
        self.threshold_ = (low_mean + high_mean) / 2
        return self

    # Predict labels for new examples.
    def predict(self, X):
        X = np.asarray(X, dtype=float)
        return (X.ravel() >= self.threshold_).astype(int)

    # Score predictions against known labels.
    def score(self, X, y):
        y = np.asarray(y, dtype=int)
        predictions = self.predict(X)

        # Return accuracy as default score.
        return float(np.mean(predictions == y))

# Create tiny training data.
X_train = np.array([[55], [62], [68], [72], [80], [85]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# Create tiny test data.
X_test = np.array([[60], [70], [78], [90]])
y_test = np.array([0, 1, 1, 1])

# Fit before predicting or scoring.
model = TinyThresholdClassifier(threshold=50.0)
model.fit(X_train, y_train)

# Predict labels for new inputs.
predicted_labels = model.predict(X_test)
test_accuracy = model.score(X_test, y_test)

# Print parameters and learned attributes.
print("Parameter threshold:", model.threshold)
print("Learned threshold_:", round(model.threshold_, 2))
print("Predictions:", predicted_labels.tolist())
print("Known labels:", y_test.tolist())
print("Score accuracy:", round(test_accuracy, 2))

```

### **2.3. Transform Methods**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_03.jpg?v=1783902177" width="250">



>* Transformers convert data into new representations
>* Fit learns rules; transform applies them

>* Fit preprocessing only on training data
>* Transform validation, test, and future data

>* Pipelines chain transformations before prediction or scoring
>* Transforms keep preprocessing consistent and reusable



```
#@title Python Code - Transform Methods

# Transform methods convert data representations.
# Fitting learns reusable transformation rules.
# Transforming applies rules without refitting.

import numpy as np

# Create tiny training and test feature matrices.
train = np.array([[60, 1.2], [72, 1.8], [66, 1.5]])
test = np.array([[69, 1.7], [63, 1.4]])

# Define a minimal standardization transformer.
class SimpleStandardScaler:
    def fit(self, data):
        self.mean_ = data.mean(axis=0)
        self.scale_ = data.std(axis=0)
        return self

    def transform(self, data):
        assert data.shape[1] == self.mean_.shape[0]
        return (data - self.mean_) / self.scale_

    def fit_transform(self, data):
        self.fit(data)
        return self.transform(data)

# Fit and transform the training data once.
scaler = SimpleStandardScaler()
train_scaled = scaler.fit_transform(train)

# Transform test data using training rules only.
test_scaled = scaler.transform(test)

# Show learned attributes and transformed summaries.
print("Learned training means:", scaler.mean_.round(2))
print("Learned training scales:", scaler.scale_.round(2))
print("Training scaled mean:", train_scaled.mean(axis=0).round(2))
print("First transformed test row:", test_scaled[0].round(2))

# Demonstrate why refitting test data changes rules.
wrong_scaler = SimpleStandardScaler()
wrong_test_scaled = wrong_scaler.fit_transform(test)

# Compare correct transform with incorrect refit.
print("Wrong test means learned:", wrong_scaler.mean_.round(2))
print("Correct first test row differs:", test_scaled[0].round(2))
print("Wrong first test row differs:", wrong_test_scaled[0].round(2))

```

## **3. Estimator Parameters**

### **3.1. Parameters Versus Attributes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_01.jpg?v=1783902179" width="250">



>* Parameters configure estimator behavior before training
>* They guide learning, not learned results

>* Attributes store what fitting learned from data
>* Do not treat learned results as settings

>* Parameters configure models before fitting
>* Attributes show learned results after fitting



```
#@title Python Code - Parameters Versus Attributes

# Parameters configure estimators before fitting.
# Attributes store results learned during fitting.
# This example separates both ideas.

# Define a tiny estimator-like class.
class TinyMeanScaler:

    # Store configuration parameters only.
    def __init__(self, with_mean=True, unit="centimeters"):
        self.with_mean = with_mean
        self.unit = unit

    # Return configurable parameters safely.
    def get_params(self):
        return {"with_mean": self.with_mean, "unit": self.unit}

    # Update parameters before fitting again.
    def set_params(self, **params):
        for name, value in params.items():
            setattr(self, name, value)
        return self

    # Learn attributes from training data.
    def fit(self, values):
        values = list(values)
        assert len(values) > 0
        self.mean_ = sum(values) / len(values)
        return self

    # Use learned attributes during transformation.
    def transform(self, values):
        assert hasattr(self, "mean_")
        if self.with_mean:
            return [round(value - self.mean_, 2) for value in values]
        return list(values)

    # Fit first, then transform immediately.
    def fit_transform(self, values):
        return self.fit(values).transform(values)

# Create small height measurements.
heights_cm = [160, 170, 180]

# Create estimator with chosen parameters.
scaler = TinyMeanScaler(with_mean=True, unit="centimeters")

# Inspect parameters before fitting.
print("Parameters before fit:", scaler.get_params())

# Fit creates learned attributes.
centered = scaler.fit_transform(heights_cm)

# Inspect learned attribute after fitting.
print("Learned attribute mean_:", scaler.mean_)

# Show transformed values briefly.
print("Centered values:", centered)

# Change a parameter, not an attribute.
scaler.set_params(with_mean=False)

# Refit after changing configuration.
unchanged = scaler.fit_transform(heights_cm)

# Compare parameter and attribute roles.
print("Updated parameters:", scaler.get_params())
print("Attribute still learned from data:", scaler.mean_)
print("Transform after parameter change:", unchanged)

```

### **3.2. Getting Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_02.jpg?v=1783902181" width="250">



>* Parameters show an estimator’s setup choices
>* They differ from learned fitted attributes

>* Reveal chosen and default model settings
>* Support collaboration, reproducibility, and explainability

>* Inspect nested pipeline and meta-estimator settings
>* Support debugging, tuning, and reproducible workflows



```
#@title Python Code - Getting Parameters

# Estimator parameters are configuration choices.
# Learned attributes appear after fitting.
# This example uses no external libraries.

# Define a tiny estimator-like teaching class.
class SimpleMeanImputer:

    # Store configuration parameters during initialization.
    def __init__(self, strategy="mean", fill_value=0):
        self.strategy = strategy
        self.fill_value = fill_value

    # Return current configuration parameters.
    def get_params(self):
        return {"strategy": self.strategy, "fill_value": self.fill_value}

    # Learn one attribute from training data.
    def fit(self, values):
        clean_values = [value for value in values if value is not None]

        # Validate that fitting has usable data.
        if len(clean_values) == 0:
            raise ValueError("At least one non-missing value is required.")

        # Choose the learned replacement value.
        if self.strategy == "mean":
            self.learned_value_ = sum(clean_values) / len(clean_values)

        # Use the configured constant otherwise.
        else:
            self.learned_value_ = self.fill_value

        return self

# Create an estimator with explicit parameters.
imputer = SimpleMeanImputer(strategy="mean", fill_value=-1)

# Inspect parameters before fitting.
params_before = imputer.get_params()
print("Parameters before fit:", params_before)

# Fit on tiny data with missing values.
training_values = [68, None, 72, 70]
imputer.fit(training_values)

# Inspect parameters after fitting.
params_after = imputer.get_params()
print("Parameters after fit:", params_after)

# Show the learned attribute separately.
print("Learned attribute:", imputer.learned_value_)

# Confirm parameters are not learned attributes.
print("Parameter keys:", list(params_after.keys()))
print("Has learned_value_ parameter:", "learned_value_" in params_after)

```

### **3.3. Updating Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_03.jpg?v=1783902183" width="250">



>* Parameters set estimator behavior before fitting
>* Updates affect future fits, not learned results

>* Parameters are settings; attributes are learned results
>* Refit after parameter changes

>* Change parameters for controlled model experiments
>* Refit before trusting learned attributes



```
#@title Python Code - Updating Parameters

# This example updates estimator parameters safely.
# Parameters are settings chosen before fitting.
# Learned attributes appear only after fitting.

# No extra installations are required.

# Define a tiny estimator-like teaching class.
class TinyMeanRegressor:

    # Store user-controlled parameters during initialization.
    def __init__(self, offset=0.0):
        self.offset = float(offset)

    # Return current configuration parameters.
    def get_params(self):
        return {"offset": self.offset}

    # Update parameters without changing learned attributes.
    def set_params(self, **params):
        if "offset" in params:
            self.offset = float(params["offset"])
        return self

    # Learn the average target value from data.
    def fit(self, X, y):
        if len(X) != len(y):
            raise ValueError("X and y need matching lengths")
        self.mean_ = sum(y) / len(y)
        return self

    # Predict using learned mean and current offset.
    def predict(self, X):
        if not hasattr(self, "mean_"):
            raise ValueError("Fit before predicting")
        return [self.mean_ + self.offset for _ in X]

# Create tiny house-size data in square feet.
X_train = [800, 1000, 1200, 1400]
y_train = [180, 220, 260, 300]

# Fit once using the initial parameter.
model = TinyMeanRegressor(offset=0.0)
model.fit(X_train, y_train)

# Inspect parameters and learned attributes separately.
print("Initial parameters:", model.get_params())
print("Learned mean_ after fit:", round(model.mean_, 1))

# Update a parameter after fitting.
model.set_params(offset=10.0)
print("Updated parameters:", model.get_params())

# The learned attribute did not change automatically.
print("mean_ after parameter update:", round(model.mean_, 1))

# Refit when new settings should match learned results.
model.fit(X_train, y_train)
print("Prediction after refit:", round(model.predict([1100])[0], 1))

```
