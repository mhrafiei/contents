>Last update: 20260712.
    
By the end of this Lecture, you will be able to:
- Distinguish estimators, predictors, transformers, and meta-estimators in Scikit-Learn workflows. 
- Use fit, predict, transform, fit_transform, and score appropriately for different estimator roles. 
- Inspect and update estimator parameters without confusing them with learned attributes. 


## **1. Estimator Roles**

### **1.1. Estimator Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_01_01.jpg?v=1783829649" width="250">



>* Estimators learn patterns from training data.
>* They include more than predictive models.

>* Before fitting, estimators are parameterized blueprints.
>* After fitting, they learn, not just predict.

>* Estimators learn at many workflow stages.
>* Shared fitting concept clarifies later roles.



### **1.2. Transformer Role**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_01_02.jpg?v=1783829659" width="250">



>* Transformers learn to reshape input data.
>* They prepare raw features for models.

>* Transformers output transformed data, not predictions.
>* They reshape features for later modeling.

>* Transformers learn and apply consistent data preparation.
>* They enable reusable, reliable pipeline workflows.



### **1.3. Meta Estimator Role**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_01_03.jpg?v=1783829651" width="250">



>* Meta-estimators coordinate and extend other estimators.
>* They structure reusable, multi-step modeling workflows.

>* Meta-estimators orchestrate components, not direct learning.
>* They unify workflows and choose best models.

>* Meta-estimators simplify reliable, reusable ML workflows.
>* They supervise models and transformations consistently.



## **2. Estimator Core Methods**

### **2.1. Fitting Estimators**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_01.jpg?v=1783829669" width="250">



>* Fit trains estimators by learning from data.
>* Learned patterns enable prediction, transformation, or evaluation.

>* Fit adds learned information to settings.
>* Results depend on the training data.

>* Fit on training data for each role.
>* Then use predict, transform, or score.



### **2.2. Using transform**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_02.jpg?v=1783829673" width="250">



>* Transform reshapes data without making predictions.
>* It creates useful features for later models.

>* Fit transformers on training data first.
>* Reuse same transform for consistent new data.

>* fit_transform trains and transforms training data
>* transformers transform features; predictors usually predict



### **2.3. Using score**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_02_03.jpg?v=1783829666" width="250">



>* Score quickly evaluates a fitted estimator.
>* Meaning varies by estimator and task.

>* Test score on unseen data.
>* Training scores can mislead model decisions.

>* Score is useful, but not sufficient.
>* Interpret scores by context and estimator.



## **3. Parameters and Attributes**

### **3.1. Parameters vs Attributes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_01.jpg?v=1783829661" width="250">



>* Parameters set behavior before training begins.
>* Attributes show estimator state, not always learned.

>* Parameters are user-set controls before training.
>* Changing them shapes model behavior and complexity.

>* Parameters are settings, not learned results.
>* Separate configuration choices from post-fit state.



### **3.2. Parameters vs Attributes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_02.jpg?v=1783829663" width="250">



>* Parameters configure estimators before training starts.
>* Users choose them, not the data.

>* Parameters are preset settings, not learned values.
>* Attributes may store training results or status.

>* Parameters set estimator behavior before training.
>* Learned attributes show training results, not settings.



### **3.3. Learned Attributes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_02/Lecture_A/image_03_03.jpg?v=1783829664" width="250">



>* Learned attributes come from fitting data.
>* Parameters are set; attributes are discovered.

>* Parameters guide training; attributes show learned results.
>* Learned attributes vary with training data.

>* Inspect attributes to understand learned patterns.
>* Change parameters, then refit; don’t edit attributes.
