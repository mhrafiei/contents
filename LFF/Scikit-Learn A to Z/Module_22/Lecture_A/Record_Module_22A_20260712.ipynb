## **2. Scalable Cluster Tuning**

### **2.1. Initialization Choices**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_02_01.jpg?v=1783855848" width="250">



>* Starting centers strongly shape K-Means results.
>* Poor initialization hurts accuracy and speed.

>* Random starts are unpredictable and sometimes misleading.
>* Separated starts improve coverage, speed, and clusters.

>* Initialization depends on data and preprocessing.
>* Thoughtful starts improve speed, stability, interpretability.



### **2.2. Initialization and Restarts**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_02_02.jpg?v=1783855846" width="250">



>* Starting centers strongly shape final clusters.
>* Good initialization improves accuracy and efficiency.

>* Multiple restarts reduce bad initialization risk.
>* Similar results increase confidence; differences suggest tuning.

>* Balance restart cost against clustering reliability.
>* Increase restarts until results become stable.



### **2.3. Selecting Cluster Count**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_02_03.jpg?v=1783855850" width="250">



>* Cluster count controls grouping detail.
>* Choose balance, not automatic truth.

>* Look for diminishing returns as clusters increase.
>* Compare inertia and silhouette, not one score.

>* Test cluster counts efficiently and check stability.
>* Choose interpretable clusters, not just better metrics.



## **3. Evaluating Mode Methods**

### **3.1. Assessing Mode Clusters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_03_01.jpg?v=1783855844" width="250">



>* Mode clustering finds dense peaks, not centers.
>* Judge real, stable, interpretable high-density groups.

>* Check fit to data shape and density.
>* Avoid over-splitting; ensure meaningful, explainable groups.

>* Visual checks reveal natural, dense cluster shapes.
>* Judge separation, outliers, and real-world usefulness.



### **3.2. Silhouette and Inertia**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_03_02.jpg?v=1783855841" width="250">



>* Silhouette still helps assess mode clusters.
>* High scores mean cohesive, separated groups.

>* Inertia fits centroid clustering better.
>* Mode clusters may look worse unfairly.

>* Read silhouette, inertia, and visuals together.
>* Visualization confirms shape when metrics mislead.



### **3.3. Mode Clustering Evaluation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_22/Lecture_A/image_03_03.jpg?v=1783855843" width="250">



>* Mode methods find dense, not spherical, groups.
>* Evaluate peaks, gaps, and irregular cluster shapes.

>* Interpret silhouette scores carefully for mode clusters.
>* Use inertia cautiously; prioritize visible data geometry.

>* Stable clusters across settings increase confidence.
>* Balance metrics, visuals, and domain knowledge.
