## **2. Training Controls**

### **2.1. Learning Rate**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_02_01.jpg?v=1783855536" width="250">



>* Learning rate sets each weight update size.
>* Good values balance training speed and stability.

>* Scaling and solver shape learning-rate behavior.
>* Adaptive rates refine training more effectively.

>* Slow or unstable training signals poor rate.
>* Tune learning rate by model behavior.



### **2.2. Batch Size**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_02_02.jpg?v=1783855530" width="250">



>* Batch size sets examples per update.
>* Small batches react faster; large batches stabilize.

>* Batch size affects memory, speed, and noise.
>* Moderate batches are a practical starting point.

>* Batch size affects convergence and generalization.
>* Tune it with learning rate, stopping.



### **2.3. Early Stopping**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_02_03.jpg?v=1783855528" width="250">



>* Stops training before overfitting begins
>* Uses validation performance to save time

>* Patience handles noisy validation score fluctuations.
>* Stop later; keep best model, save resources.

>* Validation quality determines early stopping reliability.
>* Use it alongside other training controls.



## **3. MLP Practical Limits**

### **3.1. Convergence Warning Signs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_03_01.jpg?v=1783855546" width="250">



>* Unhealthy loss curves signal poor training progress.
>* Inconsistent runs suggest unreliable convergence.

>* Poor predictions signal stalled or failed learning.
>* Chance accuracy or constant outputs are warnings.

>* Small setup changes can drastically alter training.
>* Slow, unstable training signals practical MLP limits.



### **3.2. Convergence Issues**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_03_02.jpg?v=1783855544" width="250">



>* Irregular loss surfaces can stall MLP training.
>* Poor feature scaling causes unstable weight updates.

>* Complex models and limited data hinder convergence.
>* Solver choice matters; more iterations may not.

>* Training depends on many interacting settings.
>* Check convergence through repeated careful experimentation.



### **3.3. Convergence Limits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_B/image_03_03.jpg?v=1783855548" width="250">



>* Training can stall despite continued iterations.
>* Bigger MLPs may exceed practical limits.

>* MLPs suit scaled, structured tabular problems.
>* Workflow and data quality can limit convergence.

>* Convergence does not guarantee best performance.
>* Know when another framework fits better.
