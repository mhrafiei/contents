## **1. MLP Layer Structure**

### **1.1. Input Layer**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_01_01.jpg?v=1783854560" width="250">



>* Input layer receives each example's features.
>* Its size matches the data representation.

>* Inputs must be converted into numbers.
>* Good encoding and scaling improve learning.

>* Input layer passes data, not decisions.
>* Its size shapes architecture and information flow.



```
#@title Python Code - Input Layer

# This script illustrates an MLP input layer.
# Features become input units one by one.
# Small examples make the idea concrete.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic seed.
np.random.seed(7)

# Create one house example.
feature_names = ["area_sqft", "bedrooms", "age_years", "distance_miles"]
example = np.array([1850, 3, 12, 1.8], dtype=float)

# Validate matching feature counts.
assert len(feature_names) == example.size

# Count input units directly.
input_units = example.size
print(f"One feature vector: {example.tolist()}")
print(f"Feature names: {', '.join(feature_names)}")
print(f"Input units needed: {input_units}")
print("Each input unit holds one feature value.")

# Draw a simple input layer.
y_positions = np.arange(input_units, 0, -1)
plt.figure(figsize=(5, 3.5))
plt.scatter(np.ones(input_units), y_positions, s=900, color="skyblue", edgecolors="black")

# Label each input unit.
for y, name, value in zip(y_positions, feature_names, example):
    plt.text(1, y, f"{name}\n{value}", ha="center", va="center", fontsize=9)

# Format the teaching diagram.
plt.xlim(0.6, 1.4)
plt.ylim(0.5, input_units + 0.5)
plt.xticks([])
plt.yticks([])
plt.title("Input Layer: one unit per feature")
plt.box(False)
plt.show()

```

### **1.2. Hidden Layers**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_01_02.jpg?v=1783854562" width="250">



>* Hidden layers transform inputs between input and output.
>* They learn combined features for prediction.

>* Hidden layers build simple-to-complex internal patterns.
>* They use intermediate activations to learn relationships.

>* Hidden layers control model capacity and flexibility.
>* Match network size to problem complexity.



```
#@title Python Code - Hidden Layers

# Hidden layers transform inputs into learned patterns.
# This script shows one hidden layer numerically.
# We inspect activations before final outputs.

import numpy as np

# Use a fixed seed.
rng = np.random.default_rng(7)

# Create one sample input.
x = np.array([2.0, -1.0, 0.5])

# Define weights into hidden layer.
W_hidden = np.array([[0.8, -0.4], [0.2, 0.9], [-0.5, 0.3]])
b_hidden = np.array([0.1, -0.2])

# Compute hidden layer signals.
hidden_linear = x @ W_hidden + b_hidden
hidden_activation = np.maximum(hidden_linear, 0)

# Define weights into output layer.
W_output = np.array([[1.2], [-0.7]])
b_output = np.array([0.05])

# Compute final network output.
output = hidden_activation @ W_output + b_output

# Check expected shapes safely.
assert hidden_activation.shape == (2,)
assert output.shape == (1,)

# Explain each layer briefly.
print("Input layer values:", x)
print("Hidden layer raw sums:", np.round(hidden_linear, 3))
print("Hidden layer activations:", np.round(hidden_activation, 3))
print("Output layer prediction:", np.round(output, 3))

# Show hidden units as feature detectors.
print("Hidden unit 1 mixes all three inputs.")
print("Hidden unit 2 learns a different combination.")
print("Together, hidden layers create useful intermediate representations.")

```

### **1.3. Output Layer**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_01_03.jpg?v=1783854563" width="250">



>* Final layer gives the network's prediction.
>* Output shape depends on the task.

>* Output design depends on prediction task.
>* Units and transformations match target type.

>* Output layer reveals learning success or failure.
>* Its form must match the task.



```
#@title Python Code - Output Layer

# This script explains output layer roles.
# It compares classification and regression outputs.
# Tiny examples keep results easy reading.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create hidden activations for five samples.
hidden = rng.normal(size=(5, 3))

# Confirm the hidden matrix shape.
assert hidden.shape == (5, 3)

# Build one output unit for regression.
reg_weights = np.array([[0.8], [-0.4], [0.6]])
reg_bias = np.array([2.0])

# Compute continuous regression predictions.
reg_output = hidden @ reg_weights + reg_bias

# Build three output units for classes.
cls_weights = np.array([[1.2, -0.5, 0.3],
                        [0.4, 0.9, -1.0],
                        [-0.7, 0.2, 1.1]])
cls_bias = np.array([0.1, -0.2, 0.3])

# Compute raw class scores.
cls_scores = hidden @ cls_weights + cls_bias

# Convert scores into probabilities.
shifted = cls_scores - cls_scores.max(axis=1, keepdims=True)
exp_scores = np.exp(shifted)
class_probs = exp_scores / exp_scores.sum(axis=1, keepdims=True)

# Choose the most likely class.
predicted_class = class_probs.argmax(axis=1)

# Print a short explanation.
print("Hidden layer shape:", hidden.shape)
print("Regression output shape:", reg_output.shape)
print("Classification output shape:", cls_scores.shape)
print("One regression prediction:", round(float(reg_output[0, 0]), 3))
print("One class probability row:", np.round(class_probs[0], 3))
print("Predicted classes:", predicted_class.tolist())
print("One unit suits regression; three units suit three classes.")

# Plot both output layer styles.
fig, axes = plt.subplots(1, 2, figsize=(9, 3))

# Show continuous regression outputs.
axes[0].bar(range(5), reg_output[:, 0], color="steelblue")
axes[0].set_title("Single Output Unit")
axes[0].set_xlabel("Sample")
axes[0].set_ylabel("Value")

# Show class probabilities for one sample.
axes[1].bar(["Cat", "Dog", "Bird"], class_probs[0], color="tomato")
axes[1].set_title("Three Output Units")
axes[1].set_ylabel("Probability")
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()

```

## **3. Activation and Training**

### **3.1. Activation Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_03_01.jpg?v=1783854567" width="250">



>* Nonlinear activations let MLPs learn complex patterns.
>* They shape layer responses and training refinement.

>* Activations change learning speed and signal strength.
>* They also shape sparse or smooth features.

>* Activations control gradient flow and learning speed.
>* Best activation depends on layer and task.



```
#@title Python Code - Activation Functions

# This script compares common activation functions.
# It shows outputs and gradient behavior.
# Small examples support training intuition.

import numpy as np
import matplotlib.pyplot as plt

# Use a deterministic seed here.
np.random.seed(7)

# Create a small input range.
x = np.linspace(-4.0, 4.0, 400)

# Define three common activations.
sigmoid = 1.0 / (1.0 + np.exp(-x))
tanh_y = np.tanh(x)
relu = np.maximum(0.0, x)

# Define their simple derivatives.
sigmoid_grad = sigmoid * (1.0 - sigmoid)
tanh_grad = 1.0 - tanh_y**2
relu_grad = np.where(x > 0.0, 1.0, 0.0)

# Summarize gradient strength near extremes.
edge = np.array([-4.0, 0.0, 4.0])
edge_sig = 1.0 / (1.0 + np.exp(-edge))
edge_tanh = np.tanh(edge)
edge_sig_grad = edge_sig * (1.0 - edge_sig)
edge_tanh_grad = 1.0 - edge_tanh**2
edge_relu_grad = np.where(edge > 0.0, 1.0, 0.0)

# Print a short training takeaway.
print("At x = -4, 0, 4:")
print("sigmoid gradients:", np.round(edge_sig_grad, 3))
print("tanh gradients:   ", np.round(edge_tanh_grad, 3))
print("relu gradients:   ", np.round(edge_relu_grad, 3))
print("Small edge gradients can slow backpropagation.")

# Plot activations and their gradients.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Show activation outputs.
axes[0].plot(x, sigmoid, label="sigmoid")
axes[0].plot(x, tanh_y, label="tanh")
axes[0].plot(x, relu, label="relu")
axes[0].set_title("Activation outputs")
axes[0].set_xlabel("Input")
axes[0].set_ylabel("Output")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Show gradient sizes.
axes[1].plot(x, sigmoid_grad, label="sigmoid grad")
axes[1].plot(x, tanh_grad, label="tanh grad")
axes[1].plot(x, relu_grad, label="relu grad")
axes[1].set_title("Gradient passed backward")
axes[1].set_xlabel("Input")
axes[1].set_ylabel("Derivative")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

### **3.2. Activation Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_03_02.jpg?v=1783854570" width="250">



>* Activations add nonlinearity for complex patterns.
>* They shape neuron responses and learning.

>* Sigmoid and tanh can slow learning.
>* ReLU trains faster but may deactivate neurons.

>* Hidden activations build features across layers.
>* Output activations must match the task.



```
#@title Python Code - Activation Functions

# This script compares common activation functions.
# It shows shape and gradient intuition.
# Small examples keep the lesson simple.

import numpy as np
import matplotlib.pyplot as plt

# Use a fixed seed here.
rng = np.random.default_rng(7)

# Create a small input range.
x = np.linspace(-5.0, 5.0, 400)

# Define three activation functions.
def sigmoid(values):

    return 1.0 / (1.0 + np.exp(-values))

def tanh_fn(values):

    return np.tanh(values)

def relu(values):

    return np.maximum(0.0, values)

# Define simple derivative formulas.
def sigmoid_grad(values):

    s = sigmoid(values)
    return s * (1.0 - s)

def tanh_grad(values):

    t = tanh_fn(values)
    return 1.0 - t**2

def relu_grad(values):

    return np.where(values > 0.0, 1.0, 0.0)

# Compute outputs for plotting.
y_sigmoid = sigmoid(x)
y_tanh = tanh_fn(x)

y_relu = relu(x)
g_sigmoid = sigmoid_grad(x)

g_tanh = tanh_grad(x)

g_relu = relu_grad(x)

# Check matching shapes safely.
assert x.shape == y_sigmoid.shape == g_sigmoid.shape
assert x.shape == y_tanh.shape == g_tanh.shape
assert x.shape == y_relu.shape == g_relu.shape

# Sample a few points numerically.
points = np.array([-4.0, -1.0, 0.0, 1.0, 4.0])

# Print compact training intuition.
print("At large inputs, sigmoid and tanh gradients shrink.")
print("ReLU keeps gradient 1 for positive inputs.")
print("Small gradients can slow backpropagation.")
print("Sample points:", points.tolist())

# Create one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot activation outputs first.
axes[0].plot(x, y_sigmoid, label="sigmoid")
axes[0].plot(x, y_tanh, label="tanh")
axes[0].plot(x, y_relu, label="relu")

# Label the first panel.
axes[0].set_title("Activation outputs")
axes[0].set_xlabel("Input value")
axes[0].set_ylabel("Output")
axes[0].legend()

# Plot activation gradients second.
axes[1].plot(x, g_sigmoid, label="sigmoid gradient")
axes[1].plot(x, g_tanh, label="tanh gradient")
axes[1].plot(x, g_relu, label="relu gradient")

# Label the second panel.
axes[1].set_title("Gradient passed backward")
axes[1].set_xlabel("Input value")
axes[1].set_ylabel("Gradient size")
axes[1].legend()

# Tighten layout for readability.
plt.tight_layout()
plt.show()

```

### **3.3. Backpropagation Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_03_03.jpg?v=1783854565" width="250">



>* Loss guides weight updates backward through layers.
>* It shows each connection's error responsibility.

>* Activations control how gradients flow backward.
>* Poor choices weaken learning and training stability.

>* Steady loss shows useful learning signals.
>* Backprop links errors to better predictions.



```
#@title Python Code - Backpropagation Basics

# Backpropagation sends error signals backward.
# Activations shape gradient strength strongly.
# Tiny examples make training behavior visible.

import numpy as np

# Fix randomness for repeatable learning.
rng = np.random.default_rng(7)

# Build a tiny binary dataset.
X = np.array([[0.0], [1.0]], dtype=float)
y = np.array([[0.0], [1.0]], dtype=float)

# Start with small random weights.
W1 = rng.normal(0, 0.5, size=(1, 2))
b1 = np.zeros((1, 2))
W2 = rng.normal(0, 0.5, size=(2, 1))
b2 = np.zeros((1, 1))

# Define sigmoid and its slope.
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_grad(a):
    return a * (1.0 - a)

# Define one training step.
def train_step(X, y, W1, b1, W2, b2, lr):
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    loss = np.mean((a2 - y) ** 2)
    dz2 = 2.0 * (a2 - y) * sigmoid_grad(a2)
    dW2 = a1.T @ dz2 / len(X)
    db2 = dz2.mean(axis=0, keepdims=True)

    da1 = dz2 @ W2.T
    dz1 = da1 * sigmoid_grad(a1)
    dW1 = X.T @ dz1 / len(X)
    db1 = dz1.mean(axis=0, keepdims=True)

    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

    grad_hidden = float(np.mean(np.abs(dz1)))
    grad_output = float(np.mean(np.abs(dz2)))
    return W1, b1, W2, b2, loss, grad_hidden, grad_output

# Check shapes before training.
assert X.shape == (2, 1) and y.shape == (2, 1)

# Train briefly and record behavior.
losses = []
hidden_grads = []
output_grads = []
for _ in range(2000):

    result = train_step(X, y, W1, b1, W2, b2, 1.0)
    W1, b1, W2, b2, loss, gh, go = result
    losses.append(loss)
    hidden_grads.append(gh)
    output_grads.append(go)

# Run one final forward pass.
a1 = sigmoid(X @ W1 + b1)
preds = sigmoid(a1 @ W2 + b2)

# Print a compact training summary.
print(f"Initial loss: {losses[0]:.4f}")
print(f"Final loss:   {losses[-1]:.4f}")
print(f"Output gradient mean: {np.mean(output_grads):.4f}")
print(f"Hidden gradient mean: {np.mean(hidden_grads):.4f}")

# Show learned predictions clearly.
print(f"Prediction for 0 -> {preds[0,0]:.3f}")
print(f"Prediction for 1 -> {preds[1,0]:.3f}")
print("Backpropagation reduced loss by updating both layers.")

```
