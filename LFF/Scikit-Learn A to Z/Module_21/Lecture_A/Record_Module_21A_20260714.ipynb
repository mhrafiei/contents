### **1.3. Activation Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_01_03.jpg?v=1784018334" width="250">

>* Hidden activations transform internal layer information

```
#@title Python Code - Activation Functions

# This script compares hidden activation functions.
# It shows how each function reshapes signals.
# The plot reveals different nonlinear transformations.

import numpy as np
import matplotlib.pyplot as plt

# Create simple input signals for hidden neurons.
z_values = np.linspace(-4, 4, 200)

# Compute three common scikit-learn hidden activations.
relu_values = np.maximum(0, z_values)
tanh_values = np.tanh(z_values)
logistic_values = 1 / (1 + np.exp(-z_values))

# Validate that all curves have matching lengths.
if len(z_values) != len(relu_values):
    raise ValueError("Activation arrays must match the input length.")

# Print a compact reminder about the MLP setting.
print("scikit-learn version: imported through sklearn is not needed here")
print("Hidden activations reshape signals before the output layer.")
print("ReLU keeps positives, tanh centers values, logistic compresses to 0-1.")

# Plot the activation curves on one shared axis.
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(z_values, relu_values, label="relu")
ax.plot(z_values, tanh_values, label="tanh")
ax.plot(z_values, logistic_values, label="logistic")

# Label the plot for beginner interpretation.
ax.set_title("Common Hidden-Layer Activation Functions")
ax.set_xlabel("Neuron input before activation")
ax.set_ylabel("Neuron output after activation")
ax.legend()

plt.show()

```

### **2.2. Classification Loss**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_02_02.jpg?v=1784018343" width="250">



>* Loss measures prediction probability errors
>* Optimizers adjust weights to improve probabilities

>* Confident wrong predictions receive larger penalties
>* Loss guides probability calibration and weight updates

>* Scale features for steadier MLP learning
>* Judge loss with validation and error costs



```
#@title Python Code - Classification Loss

# This example shows MLP classification loss.
# Scaling helps the optimizer use loss effectively.
# The plot shows loss decreasing during training.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load a small packaged binary classification dataset.
data = load_breast_cancer()
features = data.data
labels = data.target

# Validate the basic shape before modeling.
if features.shape[0] != labels.shape[0]:
    raise ValueError("Feature rows must match label count.")

# Split data so scaling is learned only from training rows.
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.25, stratify=labels, random_state=42
)

# Build one scaled MLP classifier for probability-based learning.
model = make_pipeline(
    StandardScaler(),
    MLPClassifier(hidden_layer_sizes=(12,), max_iter=300, random_state=42)
)

# Fit the model once and evaluate held-out predictions.
model.fit(X_train, y_train)
predicted_labels = model.predict(X_test)
test_accuracy = accuracy_score(y_test, predicted_labels)

# Read the loss curve stored by the trained MLP step.
mlp_step = model.named_steps["mlpclassifier"]
loss_values = np.array(mlp_step.loss_curve_)

print("scikit-learn version:", sklearn.__version__)
print("Final training loss:", round(float(loss_values[-1]), 4))
print("Test accuracy:", round(float(test_accuracy), 3))

# Plot how classification loss changes across training iterations.
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(range(1, len(loss_values) + 1), loss_values, color="tab:blue")
ax.set_title("MLP Classification Loss During Training")
ax.set_xlabel("Training iteration")
ax.set_ylabel("Classification loss")
plt.show()

```

### **2.3. Regression Loss**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_02_03.jpg?v=1784018340" width="250">



>* Regression MLPs predict continuous numeric values
>* Squared errors guide weight updates

>* Scale features for smoother MLP regression training
>* Transform skewed targets to improve optimization

>* Low loss may still overfit training data
>* Check validation, residuals, and real-world impact



```
#@title Python Code - Regression Loss

# This example trains a small MLP regressor.
# Scaling helps the optimizer reduce regression loss.
# The output compares loss and test error.

import numpy as np
import sklearn
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Create a small regression dataset with uneven feature scales.
features, target = make_regression(
    n_samples=500,
    n_features=4,
    noise=18.0,
    random_state=42,
)

# Stretch two columns to imitate real features with different units.
features = features.copy()
features[:, 0] = features[:, 0] * 1000.0
features[:, 1] = features[:, 1] * 0.01

# Check that the feature matrix and target match.
if features.shape[0] != target.shape[0]:
    raise ValueError("Features and target must have matching rows.")

# Split before scaling to avoid leaking test information.
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=42,
)

# Fit one MLP regressor inside a scaling pipeline.
model = make_pipeline(
    StandardScaler(),
    MLPRegressor(
        hidden_layer_sizes=(20,),
        activation="relu",
        solver="adam",
        max_iter=800,
        random_state=42,
        verbose=False,
    ),
)

model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Compare scikit-learn training loss with test mean squared error.
mlp_step = model.named_steps["mlpregressor"]
train_loss = mlp_step.loss_
test_mse = mean_squared_error(y_test, predictions)

print("scikit-learn version:", sklearn.__version__)
print("Training loss from MLPRegressor:", round(train_loss, 2))
print("Test mean squared error:", round(test_mse, 2))
print("First three predictions:", np.round(predictions[:3], 1).tolist())

```

### **3.2. MLP Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_03_02.jpg?v=1784018329" width="250">



>* MLPs transform features into class scores
>* Hidden layers learn nonlinear signal combinations

>* Output scores become class decisions
>* Confidence depends on training data quality

>* Loss helps classifiers learn from mistakes
>* Evaluate accuracy, confusion, calibration, and impact



```
#@title Python Code - MLP Classification

# This example trains a small MLP classifier.
# It shows probabilities behind class predictions.
# The output compares confidence with actual labels.

import numpy as np
import sklearn
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Load a small built-in multiclass classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic shape before modeling.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Split data so evaluation uses examples unseen during training.
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.3,
    stratify=target,
    random_state=42,
)

# Scale features before fitting the MLP classifier.
model = make_pipeline(
    StandardScaler(),
    MLPClassifier(hidden_layer_sizes=(8,), max_iter=1200, random_state=42),
)

# Fit once, then ask for labels and class probabilities.
model.fit(X_train, y_train)
predicted_labels = model.predict(X_test)
predicted_probabilities = model.predict_proba(X_test)

# Select a few examples to inspect prediction behavior.
example_indices = np.array([0, 1, 2, 3, 4])
example_probabilities = np.round(predicted_probabilities[example_indices], 2)

# Print concise results that connect probabilities to labels.
print("scikit-learn version:", sklearn.__version__)
print("Test accuracy:", round(accuracy_score(y_test, predicted_labels), 3))
print("Class names:", list(iris.target_names))
print("True labels:", list(iris.target_names[y_test[example_indices]]))
print("Predicted labels:", list(iris.target_names[predicted_labels[example_indices]]))
print("Predicted probabilities:", example_probabilities.tolist())

```

### **3.3. MLP Regression**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_21/Lecture_A/image_03_03.jpg?v=1784018332" width="250">



>* Predicts continuous values from transformed features
>* Outputs unrestricted numerical estimates, not probabilities

>* Loss reduces prediction errors during training
>* Outliers require scaling and validation checks

>* MLPs interpolate better than they extrapolate
>* Treat predictions as contextual estimates



```
#@title Python Code - MLP Regression

# This example trains a small MLP regressor.
# Scaling helps neural networks learn stable mappings.
# Predictions are compared with true continuous targets.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Load a small packaged regression dataset.
diabetes = load_diabetes()
features = diabetes.data
target = diabetes.target

# Validate the basic regression data shape.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target values.")

# Split data before scaling to avoid leakage.
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=42,
)

# Build one pipeline with scaling and an MLP regressor.
model = make_pipeline(
    StandardScaler(),
    MLPRegressor(
        hidden_layer_sizes=(20,),
        activation="relu",
        solver="lbfgs",
        alpha=0.01,
        max_iter=1000,
        random_state=42,
    ),
)

# Fit the model and predict continuous values.
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Measure typical prediction error in target units.
mae = mean_absolute_error(y_test, predictions)
print("scikit-learn version:", sklearn.__version__)
print("Test MAE in disease progression units:", round(mae, 1))
print("First prediction versus true value:", round(predictions[0], 1), round(y_test[0], 1))

# Plot predicted values against true values.
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(y_test, predictions, alpha=0.75, label="Test cases")

# Add a reference line for perfect predictions.
low_value = min(y_test.min(), predictions.min())
high_value = max(y_test.max(), predictions.max())
ax.plot([low_value, high_value], [low_value, high_value], color="black", label="Perfect")

# Label the plot for interpretation.
ax.set_title("MLP Regression: Predicted vs True Values")
ax.set_xlabel("True target value")
ax.set_ylabel("Predicted target value")
ax.legend()
plt.show()

```
