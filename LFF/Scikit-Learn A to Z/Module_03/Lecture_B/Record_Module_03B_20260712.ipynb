### **1.1. Manifold Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_01.jpg?v=1783830975" width="250">



>* High-dimensional data can follow hidden surfaces.
>* Synthetic manifolds reveal nonlinear structure and neighborhoods.

>* Manifolds connect simple examples to real data.
>* They reveal hidden geometry affecting distance interpretation.

>* Control manifolds to model noise and complexity.
>* Reveal hidden structure beyond visible coordinates.



```
#@title Python Code - Manifold Data

# This example creates curved manifold datasets.
# It shows hidden low dimensional structure.
# One plot compares two classic manifolds.

# Import plotting and dataset tools.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_s_curve, make_swiss_roll

# Set a deterministic random seed.
np.random.seed(7)

# Generate a small S curve dataset.
X_s, t_s = make_s_curve(
    n_samples=500, noise=0.05, random_state=7
)

# Generate a small Swiss roll dataset.
X_r, t_r = make_swiss_roll(
    n_samples=500, noise=0.08, random_state=7
)

# Validate expected dataset shapes.
assert X_s.shape == (500, 3)
assert X_r.shape == (500, 3)

# Print short summaries for learners.
print("S-curve shape:", X_s.shape)
print("Swiss roll shape:", X_r.shape)
print("Observed dimensions:", X_s.shape[1])
print("Noise levels: 0.05 and 0.08")

# Create one figure with two panels.
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(1, 2, 1, projection="3d")
ax2 = fig.add_subplot(1, 2, 2, projection="3d")

# Plot the S curve manifold.
ax1.scatter(
    X_s[:, 0], X_s[:, 1], X_s[:, 2], c=t_s, s=12, cmap="viridis"
)
ax1.set_title("S-curve manifold")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("z")

# Plot the Swiss roll manifold.
ax2.scatter(
    X_r[:, 0], X_r[:, 1], X_r[:, 2], c=t_r, s=12, cmap="plasma"
)
ax2.set_title("Swiss roll manifold")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("z")

# Add a clear overall title.
fig.suptitle("Manifold data: curved surfaces in 3D")

# Improve spacing for readability.
plt.tight_layout()
plt.show()

```

```
#@title Python Code - Manifold Learning Data

# Explore synthetic manifold learning data.
# Use simple geometry and clear visuals.
# Keep the example small and deterministic.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll

# Set a deterministic random seed.
np.random.seed(7)

# Generate a small swiss roll dataset.
X, t = make_swiss_roll(
    n_samples=500, noise=0.08, random_state=7
)

# Validate the generated data shape.
assert X.shape == (500, 3)
assert t.shape == (500,)

# Add one irrelevant extra feature.
extra = np.random.normal(size=(X.shape[0], 1))
X_high = np.hstack((X, extra))

# Report the key dataset properties.
print(f"Samples: {X_high.shape[0]}")
print(f"Visible dimensions: {X_high.shape[1]}")
print("Underlying structure: a curved 2D surface")
print("Extra feature added: 1 noisy dimension")

# Plot the manifold in three dimensions.
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Color points by manifold position.
ax.scatter(
    X[:, 0], X[:, 1], X[:, 2],
    c=t, cmap="viridis", s=14
)

# Add beginner-friendly plot labels.
ax.set_title("Swiss Roll: Manifold Learning Data")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Feature 3")

# Show the final visualization.
plt.tight_layout()
plt.show()

```

### **2.2. Noise and Separation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_02.jpg?v=1783830980" width="250">



>* Noise changes realism and prediction certainty.
>* Class separation controls group overlap difficulty.

>* Tune noise and separation for realistic practice.
>* They reveal sensitivity to randomness and overlap.

>* Robust models handle noise and overlap.
>* Tuning reveals realism and learning limits.



```
#@title Python Code - Noise and Separation

# Explore noise and separation visually.
# This example uses only NumPy.
# It compares easier and harder classes.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Build one synthetic classification dataset.
def make_data(samples, separation, noise_scale):
    half = samples // 2

    # Define two class centers.
    center_a = np.array([-separation, 0.0])
    center_b = np.array([ separation, 0.0])

    # Sample points around each center.
    class_a = rng.normal(center_a, noise_scale, size=(half, 2))
    class_b = rng.normal(center_b, noise_scale, size=(half, 2))

    # Combine features and labels.
    features = np.vstack((class_a, class_b))
    labels = np.array([0] * half + [1] * half)

    return features, labels

# Create two tuning scenarios.
settings = [
    (120, 2.5, 0.45, "Low noise, high separation"),
    (120, 0.9, 1.10, "High noise, low separation"),
]

# Prepare one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Draw each scenario side by side.
for ax, (n, sep, noise, title) in zip(axes, settings):
    X, y = make_data(n, sep, noise)

    # Check the generated shape safely.
    assert X.shape == (n, 2) and y.shape == (n,)

    # Plot both classes clearly.
    ax.scatter(X[y == 0, 0], X[y == 0, 1], s=35, label="Class 0")
    ax.scatter(X[y == 1, 0], X[y == 1, 1], s=35, label="Class 1")

    # Add labels and limits.
    ax.set_title(title)
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.legend(frameon=False)
    ax.grid(alpha=0.25)

# Add one short summary.
print("Left: easier pattern. Right: harder pattern from more noise and less separation.")

# Show the final comparison plot.
plt.tight_layout()
plt.show()

```

### **2.3. Tuning Data Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_03.jpg?v=1783830981" width="250">



>* Larger samples reveal patterns more reliably.
>* More features increase realism and complexity.

>* Noise adds realism through random variation.
>* Balance separation and noise for useful challenge.

>* Tune datasets to match teaching goals.
>* Iterate settings to understand model behavior.



```
#@title Python Code - Tuning Data Parameters

# This script tunes synthetic dataset parameters.
# It compares noise, size, and dimensions.
# One plot summarizes the parameter effects.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Build a simple generator function.
def make_data(samples, dims, noise):

    x = rng.normal(size=(samples, dims))
    signal = x[:, 0] * 2.0 + x[:, 1] * -1.0
    y = signal + rng.normal(scale=noise, size=samples)
    return x, y, signal

# Define small tuning scenarios.
settings = [
    (40, 2, 0.2, "small, low noise"),
    (120, 2, 0.2, "more samples"),
    (120, 2, 1.2, "more noise"),
    (120, 6, 0.2, "more dimensions"),
]

# Prepare one compact figure.
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes = axes.ravel()

# Plot each tuning scenario.
for ax, item in zip(axes, settings):

    samples, dims, noise, label = item
    x, y, signal = make_data(samples, dims, noise)
    corr = np.corrcoef(signal, y)[0, 1]

    # Show observed target versus main feature.
    ax.scatter(x[:, 0], y, s=18, alpha=0.75)
    ax.set_title(label, fontsize=10)
    ax.set_xlabel(f"feature 1 | n={samples}")
    ax.set_ylabel(f"target | d={dims}")

    # Add a short quality note.
    ax.text(
        0.03, 0.92, f"noise={noise}\ncorr={corr:.2f}",
        transform=ax.transAxes, va="top", fontsize=9
    )

# Keep the layout readable.
plt.tight_layout()

# Print a short teaching summary.
print("Higher noise weakens the visible relationship.")
print("More samples make the overall pattern steadier.")
print("More dimensions add complexity beyond the plotted view.")
plt.show()

```

### **3.1. OpenML Data Loading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_03_01.jpg?v=1783830967" width="250">



>* OpenML provides reusable datasets with helpful metadata.

>* OpenML suits benchmarks and shared experiments.
>* Inspect datasets for quality and meaning.

>* Check versions, access, and dataset preparation.
>* Thoughtful loading supports reproducible, responsible machine learning.



```
#@title Python Code - OpenML Data Loading

# This example loads a small OpenML dataset.
# It shows features, targets, and metadata.
# Network issues are handled safely here.
# pip install scikit-learn

# Import tools for loading and previewing data.
import pandas as pd
from sklearn.datasets import fetch_openml

# Try loading a tiny public dataset.
try:
    dataset = fetch_openml(
        name="iris",
        version=1,
        as_frame=True,
        parser="auto",
    )

    # Extract the feature table and target column.
    X = dataset.data
    y = dataset.target

    # Check that the dataset loaded correctly.
    if X.shape[0] > 0 and X.shape[1] > 0:
        print("Dataset name:", dataset.details.get("name", "unknown"))
        print("Samples and features:", X.shape)
        print("Target name:", dataset.target_names)
        print("Feature columns:", list(X.columns))
        print("First target labels:", list(pd.Series(y).unique()[:3]))
        print("Missing values total:", int(X.isna().sum().sum()))
    else:
        print("Dataset loaded, but shape was unexpected.")

# Fall back safely if internet access fails.
except Exception as error:
    print("OpenML load skipped:", type(error).__name__)
    print("This example needs internet access.")

```

```
#@title Python Code - Sample Images

# This example loads built in sample images.
# It inspects image shapes and pixel ranges.
# One plot shows color and grayscale views.

# Import tools for loading and plotting images.
from sklearn.datasets import load_sample_images
import matplotlib.pyplot as plt
import numpy as np

# Load the small packaged image collection.
bunch = load_sample_images()
images = bunch.images
filenames = bunch.filenames

# Validate that images were loaded safely.
assert len(images) >= 2, "Need at least two images"
assert images[0].ndim == 3, "Expected color image array"

# Select one image for inspection.
image = images[0]
height, width, channels = image.shape
pixel_min, pixel_max = int(image.min()), int(image.max())

# Convert the image to grayscale.
grayscale = image.mean(axis=2)
assert grayscale.shape == (height, width), "Unexpected grayscale shape"

# Print a short summary.
print("Loaded images:", len(images))
print("First file:", filenames[0].split("/")[-1])
print("Color shape:", image.shape)
print("Gray shape:", grayscale.shape)
print("Pixel range:", pixel_min, "to", pixel_max)

# Create one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(image)
axes[0].set_title("Color image")
axes[0].axis("off")

# Show the grayscale version beside it.
axes[1].imshow(grayscale, cmap="gray")
axes[1].set_title("Grayscale view")
axes[1].axis("off")

# Tight layout keeps labels readable.
plt.tight_layout()
plt.show()

```
