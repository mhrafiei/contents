## **1. Synthetic Supervision**

### **1.1. Classification Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_01.jpg?v=1783904469" width="250">



>* Synthetic labels teach category prediction
>* Control samples, features, classes, and separation

>* Control class overlap to change difficulty
>* Mix useful, repeated, and noisy features

>* Explore class balance and evaluation limits
>* Test classifiers on varied decision boundaries



```
#@title Python Code - Classification Data

# Generate simple synthetic classification data.
# Visualize classes using two features.
# Keep output short and beginner friendly.

import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random generator.
rng = np.random.default_rng(42)

# Choose small class sizes.
class_zero_count = 60
class_one_count = 60

# Define two class centers.
center_zero = np.array([2.0, 2.0])
center_one = np.array([5.0, 4.5])

# Generate two compact feature clouds.
class_zero = rng.normal(center_zero, 0.7, (class_zero_count, 2))
class_one = rng.normal(center_one, 0.8, (class_one_count, 2))

# Combine features and labels.
features = np.vstack([class_zero, class_one])
labels = np.array([0] * class_zero_count + [1] * class_one_count)

# Validate the generated dataset shape.
assert features.shape == (120, 2)
assert labels.shape == (120,)

# Print a compact dataset summary.
print("Synthetic classification dataset")
print(f"Rows: {features.shape[0]}, features: {features.shape[1]}")
print(f"Class 0 count: {(labels == 0).sum()}")
print(f"Class 1 count: {(labels == 1).sum()}")

# Plot the two generated classes.
plt.figure(figsize=(6, 4))
plt.scatter(features[:, 0], features[:, 1], c=labels, cmap="coolwarm", edgecolor="black")

# Add readable plot labels.
plt.title("Generated Classification Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Display one clear teaching plot.
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Regression Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_02.jpg?v=1783904471" width="250">



>* Predict continuous values from input features
>* Control patterns to study model behavior

>* Control signal and noise in regression data
>* Test model limits under uncertainty

>* Test useful, weak, and irrelevant features
>* Match model complexity to data patterns



```
#@title Python Code - Regression Data

# Synthetic regression data supports supervised learning.
# Signal and noise are controlled here.
# One plot reveals the generated relationship.

import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random number generator.
rng = np.random.default_rng(42)

# Generate one feature representing house size.
size_sqft = rng.uniform(500, 3000, 80)

# Create a smooth signal plus random noise.
true_price = 50_000 + 180 * size_sqft
noise = rng.normal(0, 35_000, size_sqft.size)
price_usd = true_price + noise

# Validate matching feature and target lengths.
assert size_sqft.shape == price_usd.shape
assert size_sqft.size == 80

# Summarize the generated regression dataset.
print(f"Samples: {size_sqft.size}")
print(f"Feature: house size in square feet")
print(f"Target: price in US dollars")
print(f"Noise standard deviation: about $35,000")

# Draw the noisy observations and true signal.
plt.figure(figsize=(7, 4))
plt.scatter(size_sqft, price_usd, alpha=0.75, label="Generated homes")
plt.plot(size_sqft, true_price, color="red", label="True signal")

# Label the regression geometry clearly.
plt.title("Synthetic Regression Data")
plt.xlabel("House size, square feet")
plt.ylabel("Price, US dollars")
plt.legend()

# Display exactly one teaching plot.
plt.tight_layout()
plt.show()

```

### **1.3. Low Rank Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_03.jpg?v=1783904473" width="250">



>* Many features, few hidden drivers
>* Controlled experiments with compact structure

>* Study models with shared hidden feature causes
>* Test learning despite many redundant features

>* Noise reveals model generalization differences
>* Compact structure simplifies high-dimensional learning



```
#@title Python Code - Low Rank Data

# Low rank data hides compact structure.
# Many features share few latent causes.
# This example visualizes synthetic supervision.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random generator.
rng = np.random.default_rng(42)

# Set small synthetic dataset dimensions.
n_samples, n_features, rank = 160, 12, 2

# Validate dimensions before generating arrays.
assert rank < n_features < n_samples

# Generate hidden low dimensional factors.
latent = rng.normal(size=(n_samples, rank))

# Mix hidden factors into many observed features.
weights = rng.normal(size=(rank, n_features))

# Add small measurement noise to features.
noise = rng.normal(scale=0.15, size=(n_samples, n_features))

# Build observed high dimensional data.
X = latent @ weights + noise

# Create a supervised target from hidden factors.
y = 3.0 * latent[:, 0] - 1.5 * latent[:, 1]

# Add small target noise for realism.
y = y + rng.normal(scale=0.4, size=n_samples)

# Estimate feature variance directions using SVD.
centered = X - X.mean(axis=0, keepdims=True)

# Compute singular values without training models.
_, singular_values, _ = np.linalg.svd(centered, full_matrices=False)

# Convert singular values into explained variance ratios.
variance = singular_values**2 / np.sum(singular_values**2)

# Print compact summary under output limits.
print(f"Observed shape: {X.shape}")
print(f"Hidden rank used: {rank}")
print(f"Top two variance share: {variance[:2].sum():.2%}")
print(f"Target range: {y.min():.2f} to {y.max():.2f}")

# Project data onto first two observed features.
fig, ax = plt.subplots(figsize=(6, 4))

# Color points by supervised target value.
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", s=35)

# Label the low rank visualization.
ax.set_title("Low Rank Synthetic Supervision")
ax.set_xlabel("Observed feature 1")
ax.set_ylabel("Observed feature 2")

# Add colorbar showing target values.
fig.colorbar(scatter, ax=ax, label="Target value")

# Display the single teaching plot.
plt.show()

```

## **2. Synthetic Data Geometry**

### **2.1. Blob Clusters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_01.jpg?v=1783904449" width="250">



>* Clusters form around central feature-space points
>* Position and spread reveal dataset structure

>* Cluster spacing and spread shape visual separation
>* Overlap reveals uncertainty in real datasets

>* Blob shape affects algorithm performance
>* Inspect geometry before trusting model results



```
#@title Python Code - Blob Clusters

# Blob clusters show simple synthetic geometry.
# We generate points around chosen centers.
# Colors reveal separation and overlap visually.

import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random number generator.
rng = np.random.default_rng(seed=42)

# Define cluster centers in feature space.
centers = np.array([[0, 0], [4, 1], [2, 4]])

# Define different spreads for each cluster.
spreads = np.array([0.45, 0.75, 1.05])

# Choose a small number of points.
points_per_cluster = 80

# Store generated points and labels.
all_points = []
all_labels = []

# Generate one Gaussian blob per center.
for label, center in enumerate(centers):
    noise = rng.normal(size=(points_per_cluster, 2))
    blob = center + noise * spreads[label]
    all_points.append(blob)
    all_labels.extend([label] * points_per_cluster)

# Combine lists into compact arrays.
X = np.vstack(all_points)
y = np.array(all_labels)

# Validate the generated dataset shape.
assert X.shape == (240, 2)
assert y.shape == (240,)

# Print a short summary for learners.
print("Generated 240 points in three blob clusters.")
print("Larger spread creates more overlap between groups.")

# Create one plot showing blob geometry.
fig, ax = plt.subplots(figsize=(7, 5))

# Plot points colored by their true cluster.
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", s=35, alpha=0.8)

# Mark the intended cluster centers clearly.
ax.scatter(centers[:, 0], centers[:, 1], c="red", marker="X", s=180)

# Add labels and a readable title.
ax.set_title("Synthetic Blob Clusters With Different Spreads")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")

# Add a simple grid for geometry reading.
ax.grid(True, alpha=0.3)
plt.show()

```

### **2.2. Moons and Circles**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_02.jpg?v=1783904455" width="250">



>* Moons and circles show nonlinear data shapes
>* Straight-line models can miss visible patterns

>* Noise blurs moons and circles
>* Patterns remain despite overlap and outliers

>* Compare algorithms on curved data shapes
>* Build intuition for classification and clustering



```
#@title Python Code - Moons and Circles

# Synthetic shapes reveal nonlinear data geometry.
# Moons and circles challenge straight boundaries.
# Noise makes ideal patterns more realistic.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Set deterministic randomness for repeatable results.
rng = np.random.default_rng(42)

# Choose small sample sizes for quick plotting.
n_samples = 300
half_samples = n_samples // 2

# Create angles for two interleaving moons.
moon_angles = rng.uniform(0, np.pi, half_samples)
moon_noise = rng.normal(0, 0.08, (n_samples, 2))

# Build the upper crescent coordinates.
upper_x = np.cos(moon_angles)
upper_y = np.sin(moon_angles)

# Build the lower crescent coordinates.
lower_x = 1 - np.cos(moon_angles)
lower_y = 0.5 - np.sin(moon_angles)

# Combine moon coordinates and labels.
moons = np.vstack((np.c_[upper_x, upper_y], np.c_[lower_x, lower_y]))
moon_labels = np.r_[np.zeros(half_samples), np.ones(half_samples)]

# Add noise after validating the shape.
assert moons.shape == moon_noise.shape
moons = moons + moon_noise

# Create angles and radii for nested circles.
circle_angles = rng.uniform(0, 2 * np.pi, half_samples)
inner_radius = rng.normal(0.45, 0.04, half_samples)

# Create outer circle radius values.
outer_radius = rng.normal(1.0, 0.05, half_samples)
circle_radii = np.r_[inner_radius, outer_radius]

# Duplicate angles for both circle classes.
all_angles = np.r_[circle_angles, circle_angles]
circle_labels = np.r_[np.zeros(half_samples), np.ones(half_samples)]

# Convert polar circle coordinates into Cartesian coordinates.
circles = np.c_[circle_radii * np.cos(all_angles), circle_radii * np.sin(all_angles)]

# Validate generated dataset sizes before plotting.
assert moons.shape == (n_samples, 2)
assert circles.shape == (n_samples, 2)

# Create one figure with two teaching panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot moons to show curved class separation.
axes[0].scatter(moons[:, 0], moons[:, 1], c=moon_labels, cmap="coolwarm", s=18)
axes[0].set_title("Two Moons")
axes[0].set_aspect("equal")

# Plot circles to show radial class separation.
axes[1].scatter(circles[:, 0], circles[:, 1], c=circle_labels, cmap="coolwarm", s=18)
axes[1].set_title("Nested Circles")
axes[1].set_aspect("equal")

# Add shared labels and compact spacing.
for axis in axes:
    axis.set_xlabel("feature 1")
    axis.set_ylabel("feature 2")

# Print a short interpretation for learners.
print("Moons separate by curved bands, not one straight line.")
print("Circles separate mostly by radius from the center.")
print("Added noise makes both patterns more realistic.")

# Display exactly one plot for visual comparison.
plt.tight_layout()
plt.show()

```

### **2.3. Swiss Roll Manifold**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_03.jpg?v=1783904457" width="250">



>* Curled sheet with hidden two-dimensional structure
>* Straight-line distances can mislead on manifolds

>* Unwrapping reveals simpler neighborhood relationships
>* Manifold methods preserve curved data structure

>* Real data can follow hidden curved surfaces
>* Visualization reveals neighborhoods, noise, and structure



```
#@title Python Code - Swiss Roll Manifold

# Generate a Swiss roll manifold.
# Color reveals the hidden unrolled direction.
# Visualize curved geometry in three dimensions.

# No extra installations are required.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create deterministic random values.
rng = np.random.default_rng(42)
n_samples = 900
noise_scale = 0.35

# Validate the chosen sample size.
if n_samples < 10:
    raise ValueError("n_samples must be at least ten")

# Generate the spiral angle values.
t = 1.5 * np.pi * (1 + 2 * rng.random(n_samples))
height = 18 * rng.random(n_samples)
noise = noise_scale * rng.normal(size=(n_samples, 3))

# Convert the sheet into three dimensions.
x = t * np.cos(t)
y = height
z = t * np.sin(t)

# Stack coordinates into one array.
points = np.column_stack((x, y, z)) + noise
assert points.shape == (n_samples, 3)

# Print a compact dataset summary.
print(f"Swiss roll points: {points.shape[0]} samples, {points.shape[1]} features")
print("Color follows the hidden unrolled spiral coordinate.")

# Build one three dimensional plot.
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Draw points colored by manifold position.
scatter = ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=t, cmap="viridis", s=12)
ax.set_title("Swiss Roll Manifold Geometry")
ax.set_xlabel("x coordinate")

# Label remaining axes clearly.
ax.set_ylabel("height coordinate")
ax.set_zlabel("z coordinate")
fig.colorbar(scatter, ax=ax, shrink=0.65, label="unrolled position")

# Show the single teaching visualization.
plt.show()

```
