## **1. Synthetic Supervision**

### **1.1. Classification Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_01.jpg?v=1783949263" width="250">



>* Synthetic labels train supervised classifiers
>* Controlled data shows noise, imbalance, overlap

>* Control class separation and task difficulty
>* Test feature quality and model behavior

>* Compare algorithms across known data patterns
>* Test imbalance and repeat controlled experiments



```
#@title Python Code - Classification Data

# Synthetic classification data supports supervised learning.
# We control separation, noise, and imbalance.
# Small plots reveal class geometry clearly.

import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random number generator.
rng = np.random.default_rng(seed=42)

# Choose a small balanced sample size.
samples_per_class = 80
feature_count = 2

# Define two class centers in centimeters.
center_low = np.array([155.0, 55.0])
center_high = np.array([175.0, 78.0])

# Generate overlapping measurements around each center.
low_risk = rng.normal(center_low, [7.0, 8.0], (samples_per_class, feature_count))
high_risk = rng.normal(center_high, [8.0, 9.0], (samples_per_class, feature_count))

# Combine features and create class labels.
features = np.vstack([low_risk, high_risk])
labels = np.array([0] * samples_per_class + [1] * samples_per_class)

# Validate the generated supervised dataset.
assert features.shape == (160, 2)
assert labels.shape == (160,)

# Count examples in each synthetic class.
class_zero_count = int(np.sum(labels == 0))
class_one_count = int(np.sum(labels == 1))

# Print a compact summary for learners.
print("Synthetic classification dataset")
print(f"Feature matrix shape: {features.shape}")
print(f"Label vector shape: {labels.shape}")
print(f"Class counts: 0={class_zero_count}, 1={class_one_count}")
print("Features represent height centimeters and weight kilograms.")

# Plot the two generated classes.
plt.figure(figsize=(7, 5))
plt.scatter(features[labels == 0, 0], features[labels == 0, 1], label="Class 0", alpha=0.75)
plt.scatter(features[labels == 1, 0], features[labels == 1, 1], label="Class 1", alpha=0.75)

# Add labels that explain the geometry.
plt.title("Synthetic Classification Data")
plt.xlabel("Height in centimeters")
plt.ylabel("Weight in kilograms")
plt.legend()

# Display the single teaching plot.
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.2. Regression Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_02.jpg?v=1783949268" width="250">



>* Predict continuous values with synthetic examples
>* Control features, noise, and difficulty

>* Model known input-output patterns with noise
>* Test informative versus distracting features

>* Test models against known regression patterns
>* Practice workflows before real-world datasets



```
#@title Python Code - Regression Data

# Synthetic regression data supports supervised learning.
# We control features, noise, and targets.
# One plot reveals the generated relationship.

import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random number generator.
rng = np.random.default_rng(seed=42)

# Generate apartment sizes in square feet.
sizes_sqft = rng.uniform(400, 1600, 80)

# Generate noisy monthly rent targets.
noise_dollars = rng.normal(0, 180, 80)
rent_dollars = 650 + 1.85 * sizes_sqft + noise_dollars

# Validate matching feature and target lengths.
assert sizes_sqft.shape == rent_dollars.shape
assert sizes_sqft.size == 80

# Summarize the generated regression dataset.
print("Synthetic regression dataset")
print(f"Samples: {sizes_sqft.size}")
print("Feature: apartment size in square feet")
print("Target: monthly rent in dollars")
print(f"Noise standard deviation: about ${noise_dollars.std():.0f}")

# Draw one scatter plot of inputs and targets.
plt.figure(figsize=(7, 4))
plt.scatter(sizes_sqft, rent_dollars, alpha=0.75)
plt.xlabel("Apartment size (square feet)")
plt.ylabel("Monthly rent (dollars)")
plt.title("Synthetic Regression Data")
plt.grid(True, alpha=0.3)
plt.show()

```

### **1.3. Low Rank Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_03.jpg?v=1783949266" width="250">



>* Many features reflect fewer hidden factors
>* Controlled data with simpler true structure

>* Test learning from correlated features
>* Study high features, few true factors

>* Reveals hidden low-dimensional data structure
>* Controls noise, factors, and target relationships



```
#@title Python Code - Low Rank Data

# Generate low rank synthetic data.
# Reveal hidden factors visually.
# Keep outputs small and clear.

import numpy as np
import matplotlib.pyplot as plt

# Set deterministic random generator.
rng = np.random.default_rng(42)

# Choose small safe dimensions.
n_samples, n_features, rank = 160, 30, 2

# Validate the requested dimensions.
assert rank < n_features < 100
assert n_samples <= 500

# Create hidden low dimensional factors.
latent = rng.normal(size=(n_samples, rank))

# Create feature mixing weights.
weights = rng.normal(size=(rank, n_features))

# Add small measurement noise.
noise = rng.normal(scale=0.15, size=(n_samples, n_features))

# Build observed high dimensional data.
X = latent @ weights + noise

# Create a supervised target from factors.
y = (latent[:, 0] + latent[:, 1] > 0).astype(int)

# Estimate singular values for rank intuition.
singular_values = np.linalg.svd(X, compute_uv=False)

# Summarize the generated dataset.
print(f"Observed data shape: {X.shape}")
print(f"Hidden factor shape: {latent.shape}")
print(f"Class counts: {np.bincount(y).tolist()}")
print(f"Top five singular values: {singular_values[:5].round(2).tolist()}")

# Plot hidden factors colored by label.
plt.figure(figsize=(6, 4))
plt.scatter(latent[:, 0], latent[:, 1], c=y, cmap="coolwarm", s=35)
plt.title("Low Rank Data: Two Hidden Factors")
plt.xlabel("Hidden factor 1")
plt.ylabel("Hidden factor 2")
plt.grid(True, alpha=0.3)
plt.show()

```

## **2. Synthetic Data Geometry**

### **2.1. Blob Clusters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_01.jpg?v=1783949258" width="250">



>* Blobs show points grouped around centers
>* Shape and spacing affect model difficulty

>* Blobs model classes and decision boundaries
>* Hidden labels reveal distance and density

>* Scaling and features change cluster appearance
>* Blobs reveal multidimensional visualization challenges



```
#@title Python Code - Blob Clusters

# Generate simple blob clusters.
# Visualize geometry with colors.
# Compare spread and separation.

# No extra installations are required.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random generator.
rng = np.random.default_rng(42)

# Define three cluster centers.
centers = np.array([
    [-3.0, -1.0],
    [0.5, 2.5],
    [3.0, -0.5],
])

# Define different cluster spreads.
spreads = np.array([0.45, 0.75, 1.05])

# Choose a small sample size.
points_per_cluster = 70

# Validate the planned dataset size.
total_points = points_per_cluster * len(centers)
assert total_points <= 300

# Build each blob separately.
points_list = []
labels_list = []

# Generate compact Gaussian clouds.
for index, center in enumerate(centers):
    noise = rng.normal(
        loc=0.0,
        scale=spreads[index],
        size=(points_per_cluster, 2),
    )

    blob_points = center + noise
    points_list.append(blob_points)
    labels_list.append(np.full(points_per_cluster, index))

# Combine blobs into one dataset.
X = np.vstack(points_list)
y = np.concatenate(labels_list)

# Validate the final feature matrix.
assert X.shape == (total_points, 2)
assert y.shape == (total_points,)

# Print a short dataset summary.
print(f"Generated points: {X.shape[0]}")
print(f"Feature columns: {X.shape[1]}")
print(f"Cluster count: {len(centers)}")
print("Larger spread creates more overlap.")

# Create one clear scatter plot.
fig, ax = plt.subplots(figsize=(7, 5))

# Plot points colored by cluster.
scatter = ax.scatter(
    X[:, 0],
    X[:, 1],
    c=y,
    cmap="viridis",
    s=35,
    alpha=0.8,
)

# Mark the true cluster centers.
ax.scatter(
    centers[:, 0],
    centers[:, 1],
    c="red",
    marker="X",
    s=180,
    label="Centers",
)

# Add labels and visual guidance.
ax.set_title("Synthetic Blob Clusters")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.grid(True, alpha=0.3)

# Add a compact legend.
ax.legend(loc="best")
plt.tight_layout()
plt.show()

```

### **2.2. Moons and Circles**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_02.jpg?v=1783949260" width="250">



>* Moons and circles reveal nonlinear data shapes
>* Straight-line models struggle with curved boundaries

>* Closeness depends on the data shape
>* Noise and spacing affect pattern difficulty

>* Compare algorithms on nonlinear shapes
>* Use plots to diagnose model assumptions



```
#@title Python Code - Moons and Circles

# Synthetic shapes reveal nonlinear class geometry.
# Moons and circles challenge straight boundaries.
# Small noisy samples keep plots readable.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random generator.
rng = np.random.default_rng(42)

# Set a small sample count.
n_samples = 300
half_samples = n_samples // 2

# Build two curved moon bands.
angles = rng.uniform(0, np.pi, half_samples)
upper_moon = np.column_stack((np.cos(angles), np.sin(angles)))
lower_moon = np.column_stack((1 - np.cos(angles), -np.sin(angles) - 0.45))
moon_points = np.vstack((upper_moon, lower_moon))

# Add gentle noise to moons.
moon_noise = rng.normal(0, 0.08, moon_points.shape)
moon_points = moon_points + moon_noise
moon_labels = np.repeat([0, 1], half_samples)

# Build nested circular rings.
circle_angles = rng.uniform(0, 2 * np.pi, n_samples)
circle_labels = np.repeat([0, 1], half_samples)
radii = np.where(circle_labels == 0, 0.45, 1.0)

# Add radial noise to circles.
radii = radii + rng.normal(0, 0.04, n_samples)
circle_x = radii * np.cos(circle_angles)
circle_y = radii * np.sin(circle_angles)
circle_points = np.column_stack((circle_x, circle_y))

# Validate generated dataset shapes.
assert moon_points.shape == (n_samples, 2)
assert circle_points.shape == (n_samples, 2)
assert moon_labels.size == n_samples
assert circle_labels.size == n_samples

# Print a compact teaching summary.
print("Generated 300 moons points and 300 circles points.")
print("Moons need curved separation between interleaving bands.")
print("Circles need radial separation between nested rings.")

# Create one figure with two panels.
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot the two moons dataset.
axes[0].scatter(moon_points[:, 0], moon_points[:, 1], c=moon_labels, s=18, cmap="coolwarm")
axes[0].set_title("Two Moons")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Plot the nested circles dataset.
axes[1].scatter(circle_points[:, 0], circle_points[:, 1], c=circle_labels, s=18, cmap="coolwarm")
axes[1].set_title("Nested Circles")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

# Use equal scaling for geometry.
axes[0].set_aspect("equal", adjustable="box")
axes[1].set_aspect("equal", adjustable="box")
plt.tight_layout()
plt.show()

```

### **2.3. Swiss Roll Manifold**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_03.jpg?v=1783949262" width="250">



>* Rolled sheet appears three-dimensional
>* Hidden structure is two-dimensional

>* Nearby layers may be far along surface
>* Manifold methods preserve true neighborhood paths

>* Swiss roll connects geometry to real data.
>* Hidden structure guides visualization and modeling.



```
#@title Python Code - Swiss Roll Manifold

# Swiss roll data shows curved geometry.
# Colors reveal the hidden unrolled coordinate.
# This example uses only NumPy plotting.

# Import numerical and plotting tools.
import numpy as np
import matplotlib.pyplot as plt

# Create a deterministic random generator.
rng = np.random.default_rng(seed=42)

# Choose a small sample size.
n_samples = 900
assert 100 <= n_samples <= 2000

# Generate the roll angle values.
t = 1.5 * np.pi * (1 + 2 * rng.random(n_samples))

# Generate height values in meters.
height_meters = 12 * rng.random(n_samples)

# Add gentle noise for realism.
noise = 0.35 * rng.normal(size=(n_samples, 3))

# Build three observed coordinates.
x_coord = t * np.cos(t)
y_coord = height_meters
z_coord = t * np.sin(t)

# Combine coordinates into one array.
points = np.column_stack((x_coord, y_coord, z_coord)) + noise
assert points.shape == (n_samples, 3)

# Create one three dimensional plot.
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Color points by hidden sheet position.
scatter = ax.scatter(
    points[:, 0], points[:, 1], points[:, 2], c=t, s=12, cmap="viridis"
)

# Label axes and title clearly.
ax.set_title("Synthetic Swiss Roll Manifold")
ax.set_xlabel("x coordinate")
ax.set_ylabel("height in meters")
ax.set_zlabel("z coordinate")

# Add a colorbar for interpretation.
colorbar = fig.colorbar(scatter, ax=ax, shrink=0.65)
colorbar.set_label("hidden unrolled position")

# Print a concise teaching summary.
print("Swiss roll shape:", points.shape)
print("Observed space is 3D, but points lie near a curled 2D sheet.")
print("Nearby layers can look close while being far along the surface.")

# Display the single visualization.
plt.show()

```
