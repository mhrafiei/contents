## **1. Synthetic Supervision**

### **1.1. Classification Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_01.jpg?v=1783978922" width="250">



>* Synthetic labels teach supervised classification patterns
>* Controlled difficulty reveals algorithm behavior

>* Control features, classes, and useful signals
>* Test models under varied data conditions

>* Isolate noise, imbalance, and boundary complexity
>* Diagnose models using known labels and structure



```
#@title Python Code - Classification Data

# Generate synthetic data for classification experiments.
# Control class separation and label noise.
# Visualize how generated classes overlap.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Create a small two-feature classification dataset.
features, labels = make_classification(
    n_samples=300,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    class_sep=1.2,
    flip_y=0.08,
    random_state=42,
)

# Validate the generated feature and label shapes.
if features.shape != (300, 2):
    raise ValueError("Expected 300 rows and 2 feature columns.")

if labels.shape[0] != features.shape[0]:
    raise ValueError("Each row must have exactly one class label.")

# Count how many examples belong to each class.
class_values, class_counts = np.unique(labels, return_counts=True)
print("Synthetic classification data")
print(f"Rows: {features.shape[0]}, features: {features.shape[1]}")
print(f"Class counts: {dict(zip(class_values, class_counts))}")
print("flip_y=0.08 adds a little label noise.")

# Plot the two generated features colored by class.
fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(
    features[:, 0],
    features[:, 1],
    c=labels,
    cmap="coolwarm",
    edgecolor="black",
    linewidth=0.4,
    alpha=0.85,
)

# Label the plot so the geometry is easy to read.
ax.set_title("Generated Classification Data")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.legend(*scatter.legend_elements(), title="Class")
plt.show()

```

### **1.2. Regression Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_02.jpg?v=1783978924" width="250">



>* Predict continuous values from generated features
>* Control patterns, noise, samples, and feature effects

>* Controlled data reveals model behavior
>* Noise and complexity show overfitting risks

>* Safe practice with realistic synthetic regression tasks
>* Adjustable data for testing prediction assumptions



```
#@title Python Code - Regression Data

# Generate synthetic data for regression learning.
# Show features, target values, and noise.
# Fit a simple model and visualize predictions.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

# Create one informative feature with controlled random noise.
features, target = make_regression(
    n_samples=120, n_features=1, noise=18.0, random_state=42
)

# Validate that the generated arrays match the lesson design.
if features.shape != (120, 1) or target.shape != (120,):
    raise ValueError("Generated regression data has an unexpected shape.")

# Split data so evaluation uses examples unseen during training.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42
)

# Fit one beginner-friendly regression model to the generated data.
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the held-out test set and summarize performance.
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Generated samples: {features.shape[0]}")
print(f"Generated features: {features.shape[1]}")
print(f"Test R-squared: {r2:.2f}")

# Sort feature values so the fitted line draws cleanly.
line_x = np.linspace(features.min(), features.max(), 100).reshape(-1, 1)
line_y = model.predict(line_x)

# Plot noisy generated targets and the learned regression line.
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(features[:, 0], target, alpha=0.65, label="Generated data")
ax.plot(line_x[:, 0], line_y, color="red", label="Learned line")
ax.set_title("Synthetic Regression Data")
ax.set_xlabel("Generated feature value")
ax.set_ylabel("Continuous target value")
ax.legend()
plt.show()

```

### **1.3. Low Rank Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_01_03.jpg?v=1783978926" width="250">



>* Many features, few hidden drivers
>* Study redundancy, correlation, and hidden structure

>* Predictions rely on few hidden factors
>* Redundant features motivate dimensionality reduction

>* Test clustering without target labels
>* Control hidden factors, noise, and feature links



```
#@title Python Code - Low Rank Data

# Generate low rank synthetic supervised data.
# Show hidden factors behind many features.
# Visualize redundancy using principal components.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_low_rank_matrix
from sklearn.decomposition import PCA

# Create many observed features from few hidden factors.
features = make_low_rank_matrix(
    n_samples=300,
    n_features=20,
    effective_rank=2,
    random_state=42,
)

# Add a supervised target driven by the hidden structure.
rng = np.random.default_rng(42)
target = 3 * features[:, 0] - 2 * features[:, 1]
target = target + rng.normal(0, 0.05, size=features.shape[0])

# Validate the generated table shape before analysis.
if features.shape != (300, 20):
    raise ValueError("Expected 300 rows and 20 features.")

# PCA reveals how much variation needs few directions.
pca = PCA(n_components=2, random_state=42)
projected = pca.fit_transform(features)
explained = pca.explained_variance_ratio_.sum()

print("Synthetic data shape: 300 rows x 20 features")
print("Hidden effective rank requested: 2")
print(f"Variance captured by first 2 PCs: {explained:.1%}")

# Color points by the supervised target value.
fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(
    projected[:, 0],
    projected[:, 1],
    c=target,
    cmap="viridis",
    s=35,
)

# Label the compact two-dimensional view.
ax.set_title("Low Rank Data: 20 Features, Two Main Directions")
ax.set_xlabel("Principal component 1")
ax.set_ylabel("Principal component 2")
fig.colorbar(scatter, ax=ax, label="Synthetic target")

plt.show()

```

## **2. Synthetic Data Geometry**

### **2.1. Blob Clusters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_01.jpg?v=1783978928" width="250">



>* Clusters show groups around central points
>* Control spacing and spread to test difficulty

>* Center spacing and spread shape cluster clarity
>* Overlapping groups reflect real modeling challenges

>* Blob shapes reveal model assumptions
>* Plots build intuition about clustering decisions



```
#@title Python Code - Blob Clusters

# Generate blob clusters for visual geometry learning.
# Compare compact and spread clusters in one plot.
# Expect overlap to increase with cluster spread.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Create two blob datasets with identical centers.
centers = np.array([[-4.0, -1.0], [0.0, 3.0], [4.0, -1.0]])
compact_points, compact_labels = make_blobs(
    n_samples=180, centers=centers, cluster_std=0.55, random_state=42
)

# Create a second dataset with more dispersion.
spread_points, spread_labels = make_blobs(
    n_samples=180, centers=centers, cluster_std=1.45, random_state=42
)

# Validate that both datasets have matching two-dimensional shapes.
if compact_points.shape != spread_points.shape:
    raise ValueError("The two blob datasets should have matching shapes.")

# Shift the spread dataset right for side-by-side comparison.
spread_points_shifted = spread_points.copy()
spread_points_shifted[:, 0] = spread_points_shifted[:, 0] + 11.0

# Print a concise summary of the geometry change.
print("Compact blobs: small spread, clearer visual separation.")
print("Spread blobs: larger spread, more overlap near borders.")
print("Samples per view:", compact_points.shape[0])

# Plot both views on one axes using the same cluster colors.
fig, ax = plt.subplots(figsize=(8, 5))
scatter_one = ax.scatter(
    compact_points[:, 0], compact_points[:, 1], c=compact_labels, s=35
)

# Add the spread blobs to the same axes.
ax.scatter(
    spread_points_shifted[:, 0], spread_points_shifted[:, 1], c=spread_labels, s=35
)
ax.scatter(centers[:, 0], centers[:, 1], c="black", marker="x", s=90)

# Mark the shifted centers for the spread view.
shifted_centers = centers.copy()
shifted_centers[:, 0] = shifted_centers[:, 0] + 11.0
ax.scatter(shifted_centers[:, 0], shifted_centers[:, 1], c="black", marker="x", s=90)

# Label the two visual conditions directly on the plot.
ax.text(-4.8, 5.2, "Compact clusters", fontsize=11)
ax.text(6.2, 5.2, "Spread clusters", fontsize=11)
ax.set_title("Blob Cluster Geometry: Separation Versus Spread")

# Label axes and show the single teaching figure.
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_aspect("equal", adjustable="box")
plt.show()

```

### **2.2. Moons and Circles**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_02.jpg?v=1783978930" width="250">



>* Moons and circles reveal curved data structure
>* They show limits of linear models

>* Noise reveals uncertain curved decision boundaries
>* Synthetic shapes model real nonlinear patterns

>* Visuals reveal curved patterns models must learn
>* Flexible methods handle non-round clusters better



```
#@title Python Code - Moons and Circles

# Generate two classic nonlinear synthetic datasets.
# Compare moons and circles in one view.
# Notice curved structure beyond straight separation.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons

# Create deterministic moons and circles with matching sample sizes.
moons_points, moons_labels = make_moons(
    n_samples=300, noise=0.08, random_state=42
)

circles_points, circles_labels = make_circles(
    n_samples=300, noise=0.05, factor=0.45, random_state=42
)

# Move circles to the right so both shapes share one axes.
circles_points = circles_points + np.array([3.0, 0.0])

# Validate that both generated datasets have two numeric features.
if moons_points.shape != (300, 2) or circles_points.shape != (300, 2):
    raise ValueError("Expected 300 rows and 2 columns for each dataset.")

print("Moons shape: 300 samples, 2 features, 2 curved classes.")
print("Circles shape: 300 samples, 2 features, 2 nested classes.")
print("Noise makes the ideal curves look more like real observations.")

# Plot both datasets on one axes to compare their geometry.
fig, ax = plt.subplots(figsize=(8, 4.8))

moon_plot = ax.scatter(
    moons_points[:, 0], moons_points[:, 1], c=moons_labels, cmap="coolwarm", s=28
)

circle_plot = ax.scatter(
    circles_points[:, 0], circles_points[:, 1], c=circles_labels, cmap="viridis", s=28
)

ax.set_title("Synthetic nonlinear geometry: moons and circles")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.text(-1.2, 1.25, "Moons", fontsize=11)
ax.text(2.0, 1.25, "Circles", fontsize=11)
ax.grid(True, alpha=0.25)
plt.show()

```

### **2.3. Swiss Roll Manifold**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_02_03.jpg?v=1783978932" width="250">



>* Rolled sheet shows hidden two-dimensional structure
>* Reveals simple patterns inside complex data

>* Nearby layers may be far along surface
>* Manifold methods preserve true neighborhood relationships

>* Real data often follows curved hidden structures
>* Swiss rolls reveal simpler underlying geometry



```
#@title Python Code - Swiss Roll Manifold

# Generate a small Swiss roll dataset.
# Color reveals the hidden unrolled coordinate.
# The plot shows curved manifold geometry.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_swiss_roll

# Create deterministic points on a rolled two-dimensional surface.
points, roll_position = make_swiss_roll(
    n_samples=1200,
    noise=0.25,
    random_state=42,
)

# Validate the generated shape before plotting.
if points.shape != (1200, 3):
    raise ValueError("Expected 1200 points with three observed coordinates.")

# Use two observed coordinates to show the spiral cross-section.
x_coordinate = points[:, 0]
z_coordinate = points[:, 2]

print("Swiss roll samples: 1200")
print("Observed coordinates per point: 3")
print("Color shows position along the rolled sheet.")

# Plot one view where nearby layers can appear close.
fig, ax = plt.subplots(figsize=(7, 6))
scatter = ax.scatter(
    x_coordinate,
    z_coordinate,
    c=roll_position,
    cmap="viridis",
    s=12,
    alpha=0.85,
)

ax.set_title("Swiss Roll Manifold: Spiral Cross-Section")
ax.set_xlabel("Observed x coordinate")
ax.set_ylabel("Observed z coordinate")
fig.colorbar(scatter, ax=ax, label="Position along the sheet")
plt.show()

```

### **3.2. SVMLight File Loading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_03_02.jpg?v=1783978936" width="250">



>* Sparse text datasets store only active features
>* Scikit-learn loads SVMLight files directly

>* Loads features sparsely with separate target labels
>* Saves memory for high-dimensional model training

>* Use utilities to avoid import mistakes
>* Keep feature spaces aligned across files



```
#@title Python Code - SVMLight File Loading

# This example loads sparse SVMLight text data.
# It shows labels and nonzero feature entries.
# The result confirms a model-ready sparse matrix.

from io import BytesIO

import numpy as np
from sklearn.datasets import load_svmlight_file

# SVMLight stores each row as a label and sparse features.
svmlight_text = "1 1:0.8 3:1.5\n0 2:2.0\n1 1:0.4 2:0.7 4:1.0"

# Scikit-learn can read file-like objects from in-memory text.
svmlight_file = BytesIO(svmlight_text.encode("utf-8"))

# Loading returns a sparse feature matrix and target labels.
features, labels = load_svmlight_file(svmlight_file, n_features=4)

# Validate the expected small teaching example shape.
if features.shape != (3, 4):
    raise ValueError("Expected 3 rows and 4 feature columns.")

# Convert only this tiny example to dense form for display.
dense_features = features.toarray()
nonzero_count = features.nnz

print("Loaded SVMLight data into a sparse matrix.")
print(f"Feature matrix shape: {features.shape}")
print(f"Stored nonzero values: {nonzero_count}")
print(f"Target labels: {labels.astype(int).tolist()}")
print("Dense view for this tiny example:")
print(np.round(dense_features, 2))

```

### **3.3. OpenML Dataset Loading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_B/image_03_03.jpg?v=1783978938" width="250">



>* Load real-world datasets directly into scikit-learn
>* Compare models using shared benchmark data

>* OpenML includes data plus helpful metadata
>* Inspect targets, missing values, and feature types

>* Identify datasets consistently for reproducible experiments
>* Practice full workflows on diverse external data



```
#@title Python Code - OpenML Dataset Loading

# This example loads an OpenML dataset safely.
# It shows metadata and tabular structure.
# A packaged fallback keeps the lesson runnable.

import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.datasets import load_iris

# Try one documented OpenML request, then validate it.
try:
    openml_data = fetch_openml(data_id=61, as_frame=True, cache=True)
    features = openml_data.data.iloc[:, :4].copy()
    target = openml_data.target.copy()
    if len(features) != len(target) or features.shape[1] != 4:
        raise ValueError("Unexpected dataset shape")
    source_name = "OpenML iris dataset, data_id 61"
except Exception:
    iris = load_iris(as_frame=True)
    features = iris.data.copy()
    target = pd.Series(iris.target_names[iris.target], name="class")
    source_name = "packaged iris fallback"
    print("OpenML unavailable, using packaged iris fallback.")

# Inspect the loaded data before modeling.
row_count = features.shape[0]
column_count = features.shape[1]
missing_values = int(features.isna().sum().sum())
class_count = target.nunique()

print("Source:", source_name)
print("Feature shape:", row_count, "rows by", column_count, "columns")
print("Target classes:", class_count)
print("Missing feature values:", missing_values)
print("First feature names:", list(features.columns[:4]))

```
