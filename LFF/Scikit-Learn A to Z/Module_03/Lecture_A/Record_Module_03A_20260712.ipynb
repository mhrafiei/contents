## **1. Classic Builtin Datasets**

### **1.1. Iris Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_01.jpg?v=1783904002" width="250">



>* Small, clean dataset for classification practice
>* Predict iris species from flower measurements

>* Match flower measurements to species labels
>* Inspect dataset shape, names, and balance

>* Shows easy and overlapping class patterns
>* Best for practicing before messier datasets



```
#@title Python Code - Iris Dataset

# Inspect Iris as classification data.
# Builtin datasets support quick demonstrations.
# Small clean data helps beginners.

# No extra installations are required.

# Import dataset loader and plotting tools.
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris as a dictionary-like object.
iris_data = load_iris()

# Validate the expected rectangular feature matrix.
assert iris_data.data.shape == (150, 4)
assert iris_data.target.shape == (150,)

# Build a small labeled table for inspection.
iris_frame = pd.DataFrame(
    iris_data.data,
    columns=iris_data.feature_names,
)

# Add readable species names for each row.
iris_frame["species"] = pd.Categorical.from_codes(
    iris_data.target,
    iris_data.target_names,
)

# Print compact dataset facts only.
print("Dataset: Iris classification")
print(f"Samples and features: {iris_data.data.shape}")
print(f"Target classes: {list(iris_data.target_names)}")
print(f"Feature names: {iris_data.feature_names}")
print("Class counts:")
print(iris_frame["species"].value_counts().to_string())

# Plot two petal measurements by species.
sns.scatterplot(
    data=iris_frame,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species",
)

# Add a clear title and display plot.
plt.title("Iris petal measurements by species")
plt.show()

```

### **1.2. Diabetes Regression Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_02.jpg?v=1783904004" width="250">



>* Predicts patient disease progression as a number
>* Small dataset for practicing regression workflows

>* Standardized numerical patient and clinical features
>* Inspect shape, names, and target distribution

>* Practice regression with continuous diabetes outcomes
>* Evaluate errors and interpret predictions responsibly



```
#@title Python Code - Diabetes Regression Dataset

# Inspect a classic regression dataset.
# Diabetes targets are continuous numbers.
# Builtin data needs no files.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load the diabetes dataset.
diabetes = load_diabetes()

# Store features and target.
X = diabetes.data
y = diabetes.target

# Validate expected two dimensional features.
assert X.ndim == 2 and y.ndim == 1

# Build a small readable table.
preview = pd.DataFrame(
    X[:5], columns=diabetes.feature_names
)

# Print compact dataset facts.
print("Dataset: diabetes regression")
print(f"Samples: {X.shape[0]}, features: {X.shape[1]}")
print(f"Feature names: {diabetes.feature_names}")
print(f"Target range: {y.min():.0f} to {y.max():.0f}")

# Show only five rows.
print("First five standardized rows:")
print(preview.round(3).to_string(index=False))

# Plot the continuous target distribution.
plt.figure(figsize=(7, 4))
plt.hist(y, bins=20, color="steelblue", edgecolor="white")
plt.title("Diabetes Dataset Target Distribution")
plt.xlabel("Disease progression score")
plt.ylabel("Patient count")
plt.tight_layout()
plt.show()

```
