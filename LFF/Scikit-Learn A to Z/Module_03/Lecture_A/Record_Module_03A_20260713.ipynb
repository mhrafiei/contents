## **1. Classic Built In Datasets**

### **1.1. Iris Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_01.jpg?v=1783978037" width="250">



>* Small, clean dataset for classification practice
>* Predict iris species from four measurements

>* Inspect features, labels, rows, and columns
>* Compare measurement ranges and species separation

>* Connect inspected measurements to classifier behavior
>* Practice evaluating feature usefulness and data limits



```
#@title Python Code - Iris Dataset

# Load the Iris dataset from scikit-learn.
# Inspect features, targets, and sample measurements.
# Summarize its supervised classification structure.

from sklearn.datasets import load_iris
import pandas as pd

# Load Iris as a Bunch object with data and labels.
iris = load_iris()

# Validate the expected tabular classification structure.
if iris.data.shape != (150, 4):
    raise ValueError("Expected 150 flowers with 4 measurement features.")

if len(iris.target) != iris.data.shape[0]:
    raise ValueError("Each flower should have exactly one species label.")

# Build a small labeled table for beginner-friendly inspection.
iris_table = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_table["species"] = [iris.target_names[label] for label in iris.target]

# Show the most important dataset facts.
print("Dataset: Iris classification data")
print(f"Feature matrix shape: {iris.data.shape}")
print(f"Target array shape: {iris.target.shape}")
print(f"Feature names: {list(iris.feature_names)}")
print(f"Target classes: {list(iris.target_names)}")

# Show one compact sample from each species.
sample_rows = iris_table.groupby("species", sort=True).head(1)
sample_rows = sample_rows.round(1)
print("One example flower from each species:")
print(sample_rows.to_string(index=False))

```

### **1.2. Diabetes Regression Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_02.jpg?v=1783978039" width="250">



>* Predicts numerical diabetes progression from patient measurements
>* Shows regression for real-world continuous outcomes

>* Match patient features to numerical outcomes
>* Inspect dataset size, columns, and target range

>* Practice regression workflow with prepared numerical data
>* Recognize teaching value and clinical limits



```
#@title Python Code - Diabetes Regression Dataset

# Load the diabetes dataset for regression inspection.
# Connect feature rows with numerical target values.
# Print compact facts before any modeling.

import sklearn
from sklearn.datasets import load_diabetes

# Load the packaged dataset as a pandas-friendly object.
diabetes = load_diabetes(as_frame=True)

# Separate inputs and the continuous target.
features = diabetes.data
target = diabetes.target

# Validate the row-to-target relationship before inspection.
if len(features) != len(target):
    raise ValueError("Each feature row must have one target value.")

# Select a tiny preview for readable classroom output.
preview_columns = ["age", "sex", "bmi", "bp"]
preview = features.loc[:2, preview_columns].round(3)

# Summarize the dataset structure and target scale.
print(f"scikit-learn version: {sklearn.__version__}")
print(f"Dataset: diabetes regression")
print(f"Samples: {features.shape[0]}")
print(f"Features: {features.shape[1]}")
print(f"First feature names: {', '.join(diabetes.feature_names[:5])}")
print(f"Target type: continuous disease progression score")
print(f"Target range: {target.min():.0f} to {target.max():.0f}")
print("First 3 rows and 4 features:")
print(preview.to_string())

```

### **1.3. Wine and Cancer**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_03.jpg?v=1783978041" width="250">



>* Wine data links samples to chemical features
>* Inspect classes, feature names, and targets

>* Cancer data uses tabular tumor measurements
>* Supports binary classification with expert interpretation

>* Practice inspecting datasets before modeling
>* Compare binary and multiclass classification examples



```
#@title Python Code - Wine and Cancer

# This script loads two built-in classification datasets.
# It compares their shapes, features, and targets.
# The output highlights multiclass versus binary classification.

from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer

# Load each dataset as a scikit-learn Bunch object.
wine = load_wine()
cancer = load_breast_cancer()

# Validate that rows and targets match for both datasets.
if wine.data.shape[0] != len(wine.target):
    raise ValueError("Wine rows and targets do not match.")

if cancer.data.shape[0] != len(cancer.target):
    raise ValueError("Cancer rows and targets do not match.")

# Store the datasets with beginner-friendly display names.
datasets = [("Wine", wine), ("Breast cancer", cancer)]

# Print a compact inspection summary for each dataset.
for dataset_name, dataset in datasets:
    sample_count = dataset.data.shape[0]
    feature_count = dataset.data.shape[1]
    class_count = len(dataset.target_names)
    first_features = ", ".join(dataset.feature_names[:3])
    target_names = ", ".join(dataset.target_names)
    print(f"{dataset_name}: {sample_count} samples, {feature_count} features.")
    print(f"{dataset_name}: {class_count} target classes: {target_names}.")
    print(f"{dataset_name}: first features are {first_features}.")

# Confirm the main classification difference in one final line.
print("Wine is multiclass; breast cancer is binary classification.")

```

### **2.1. Digits Image Data**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_02_01.jpg?v=1783978046" width="250">



>* Handwritten digits stored as grayscale pixel grids
>* Pixel intensities become features for classification

>* Images appear as grids or flattened features
>* Labels identify ten possible digit classes

>* Prepared digits support concept-focused classification demos
>* Pixel locations act as image features



```
#@title Python Code - Digits Image Data

# Load handwritten digit images from scikit-learn.
# Compare image grids with flattened features.
# Display one digit and its numeric structure.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# The digits dataset is packaged with scikit-learn.
digits = load_digits()

# Each image has a matching flattened feature row.
image_count = len(digits.images)
feature_count = digits.data.shape[1]

# Validate the expected image and feature shapes.
if digits.images.shape[1:] != (8, 8):
    raise ValueError("Expected each digit image to be 8 by 8 pixels.")

if feature_count != 64:
    raise ValueError("Expected each flattened digit to have 64 features.")

# Select one observation for a focused inspection.
example_index = 0
example_image = digits.images[example_index]
example_features = digits.data[example_index]
example_label = digits.target[example_index]

# Print concise facts about the two dataset views.
print(f"Images shape: {digits.images.shape}")
print(f"Flattened data shape: {digits.data.shape}")
print(f"Target shape: {digits.target.shape}")
print(f"Example label: {example_label}")
print(f"First 10 flattened pixel values: {example_features[:10].astype(int)}")

# Show the image-shaped view of the same observation.
fig, ax = plt.subplots(figsize=(4, 4))
ax.imshow(example_image, cmap="gray_r", interpolation="nearest")
ax.set_title("One 8 by 8 handwritten digit image")
ax.set_xlabel("Pixel column")
ax.set_ylabel("Pixel row")
plt.show()

```

### **2.2. Linnerud Multioutput**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_02_02.jpg?v=1783978048" width="250">



>* Predicts multiple related body measurements
>* Uses exercise features as inputs

>* Rows align exercise features with multiple targets
>* Small dataset for teaching multioutput structure

>* Multioutput problems appear across many fields
>* One sample can have several related targets



```
#@title Python Code - Linnerud Multioutput

# Inspect Linnerud as multioutput regression data.
# Each sample has several numeric targets.
# The output confirms aligned feature and target rows.

from sklearn.datasets import load_linnerud

# Load the packaged dataset as pandas tables.
linnerud = load_linnerud(as_frame=True)
features = linnerud.data
targets = linnerud.target

# Check the row alignment before inspecting values.
if len(features) != len(targets):
    raise ValueError("Feature and target rows must match.")

# Combine one sample to show multiple targets per person.
first_sample = features.iloc[[0]].join(targets.iloc[[0]])
rounded_sample = first_sample.round(1)

print("Linnerud is a multioutput regression dataset.")
print(f"Feature matrix shape: {features.shape}")
print(f"Target matrix shape: {targets.shape}")
print(f"Feature columns: {list(features.columns)}")
print(f"Target columns: {list(targets.columns)}")
print("First aligned row, features followed by targets:")
print(rounded_sample.to_string(index=False))

```

## **3. Choosing Built In Data**

### **3.1. Classification Dataset Fit**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_03_01.jpg?v=1783978050" width="250">



>* Use classification for category prediction tasks
>* Choose categorical targets, not continuous values

>* Match dataset complexity to teaching goals
>* Use simple or rich data appropriately

>* Match class difficulty to evaluation goals
>* Consider class count, balance, and clarity



```
#@title Python Code - Classification Dataset Fit

# Choose classification data for a simple modeling demonstration.
# Compare dataset structure before fitting a classifier.
# See why iris supports beginner classification lessons.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load a small built-in classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic classification structure.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

if len(iris.target_names) < 2:
    raise ValueError("Classification needs at least two classes.")

# Split data while preserving class proportions.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42, stratify=target
)

# Fit one beginner-friendly classifier.
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# Evaluate whether the dataset supports a clear demo.
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Dataset: iris classification with {features.shape[0]} rows.")
print(f"Features: {features.shape[1]} numeric measurements.")
print(f"Classes: {', '.join(iris.target_names)}.")
print(f"Test accuracy from one simple model: {accuracy:.2f}")

```

### **3.2. Regression Dataset Fit**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_03_02.jpg?v=1783978052" width="250">



>* Use regression for continuous numeric targets
>* Evaluate predictions by error size

>* Choose understandable features learners can interpret
>* Match dataset size and focus to goals

>* Match dataset complexity to teaching goals
>* Use realistic patterns for meaningful demonstrations



```
#@title Python Code - Regression Dataset Fit

# This example chooses regression data for modeling.
# The target is a continuous numerical measurement.
# A simple model shows prediction error clearly.

import sklearn
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Load a packaged regression dataset with numeric features.
diabetes = load_diabetes(as_frame=True)
features = diabetes.data
target = diabetes.target

# Check that rows and targets match before modeling.
if len(features) != len(target):
    raise ValueError("Feature rows must match target values.")

# Split data so evaluation uses examples unseen during training.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42
)

# Fit one beginner-friendly regression model.
model = LinearRegression()
model.fit(X_train, y_train)

# Predict continuous values and measure average absolute error.
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Dataset choice: diabetes regression, {features.shape[0]} rows.")
print(f"Target type: continuous disease progression score.")
print(f"Mean absolute error on test data: {mae:.1f} score units.")

```
