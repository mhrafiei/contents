## **1. Classic Built In Datasets**

### **1.1. Iris Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_01.jpg?v=1783948682" width="250">



>* Iris uses four flower measurements
>* Predict species for classification practice

>* Features align each flower with species labels
>* Inspect features, classes, samples, and balance

>* Small dataset supports visual classification learning
>* Builds habits for inspecting data before modeling



```
#@title Python Code - Iris Dataset

# Inspect the classic Iris dataset.
# Focus on features and labels.
# Keep output short and readable.

# No extra installations are required.

# Import dataset tools and plotting.
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris as a structured object.
iris_data = load_iris()
feature_names = iris_data.feature_names
target_names = iris_data.target_names

# Build a small inspection dataframe.
iris_frame = pd.DataFrame(
    iris_data.data,
    columns=feature_names,
)

# Add readable species names.
iris_frame["species"] = pd.Categorical.from_codes(
    iris_data.target,
    target_names,
)

# Validate the expected dataset shape.
assert iris_frame.shape == (150, 5)
assert len(target_names) == 3

# Print compact dataset facts.
print("Dataset: Iris classification")
print(f"Samples: {iris_data.data.shape[0]}")
print(f"Numeric features: {iris_data.data.shape[1]}")
print(f"Feature names: {feature_names}")
print(f"Class names: {list(target_names)}")
print("Class counts:")
print(iris_frame["species"].value_counts().to_string())

# Show a tiny preview only.
print("First three rows:")
print(iris_frame.head(3).to_string(index=False))

# Plot two informative petal measurements.
sns.scatterplot(
    data=iris_frame,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species",
)

# Label the plot clearly.
plt.title("Iris species by petal measurements")
plt.xlabel("Petal length in centimeters")
plt.ylabel("Petal width in centimeters")
plt.tight_layout()
plt.show()

```

### **1.2. Diabetes Regression Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_02.jpg?v=1783948687" width="250">



>* Regression predicts numerical disease progression
>* Rows are patients, columns are features

>* Inspect standardized patient features and target values
>* Understand dataset shape before regression modeling

>* Simple dataset for regression practice
>* Interpret predictions responsibly, not clinically



```
#@title Python Code - Diabetes Regression Dataset

# Inspect a classic regression dataset.
# Focus on structure before modeling.
# Keep output short and readable.

# No extra installations are required.

# Import dataset tools and pandas.
from sklearn.datasets import load_diabetes
import pandas as pd

# Load data as a pandas frame.
diabetes = load_diabetes(as_frame=True)
features = diabetes.data
target = diabetes.target

# Validate expected row alignment.
assert len(features) == len(target)
assert features.shape[1] == len(diabetes.feature_names)

# Build a compact inspection table.
preview = features.head(3).round(3)
preview["target"] = target.head(3).astype(int)

# Print the dataset purpose.
print("Diabetes is a regression dataset.")
print(f"Rows: {features.shape[0]}, features: {features.shape[1]}")

# Print feature names compactly.
print("Feature names:", ", ".join(diabetes.feature_names))
print("Target meaning: disease progression after one year.")

# Print target range safely.
minimum = int(target.min())
maximum = int(target.max())
print(f"Target range: {minimum} to {maximum}")

# Show only three rounded records.
print("First three rows, rounded:")
print(preview.to_string(index=False))

```

### **1.3. Wine and Cancer**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_03/Lecture_A/image_01_03.jpg?v=1783948685" width="250">



>* Wine data predicts cultivars from chemical features
>* Inspect samples, features, targets, and class distribution

>* Cancer data supports binary medical classification
>* Inspect features, labels, balance, and description

>* Wine shows multiclass; cancer shows binary classification
>* Inspect structure before choosing a dataset



```
#@title Python Code - Wine and Cancer

# Compare two classic classification datasets.
# Inspect shapes, features, and targets.
# Choose datasets before modeling demonstrations.

from sklearn.datasets import load_wine, load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets as dictionary like objects.
wine = load_wine()
cancer = load_breast_cancer()

# Validate expected two dimensional feature matrices.
assert wine.data.ndim == 2 and cancer.data.ndim == 2
assert wine.target.size == wine.data.shape[0]

# Build a compact comparison table.
summary = pd.DataFrame({
    "dataset": ["wine", "cancer"],
    "samples": [wine.data.shape[0], cancer.data.shape[0]],
    "features": [wine.data.shape[1], cancer.data.shape[1]],
    "classes": [len(wine.target_names), len(cancer.target_names)],
})

# Print only compact inspection results.
print("Classic built-in classification datasets")
print(summary.to_string(index=False))
print("Wine target names:", list(wine.target_names))
print("Cancer target names:", list(cancer.target_names))

# Count examples in each target class.
wine_counts = pd.Series(wine.target).value_counts().sort_index()
cancer_counts = pd.Series(cancer.target).value_counts().sort_index()

# Create one small class balance plot.
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].bar(wine.target_names, wine_counts.values)
axes[0].set_title("Wine classes")
axes[1].bar(cancer.target_names, cancer_counts.values)

# Label the second plot clearly.
axes[1].set_title("Cancer classes")
fig.suptitle("Class balance in built-in datasets")
plt.tight_layout()
plt.show()

```
