## **2. Sampling Schemes**

### **2.1. Bootstrap Sampling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_B/image_02_01.jpg?v=1783850194" width="250">



>* Samples with replacement create varied training sets.
>* Diverse models improve bagging stability and robustness.

>* Resampling creates varied training sets.
>* Averaging models improves generalization reliability.

>* Bootstrap balances model diversity and dataset fidelity.
>* Omitted cases help check robustness and uncertainty.



```
#@title Python Code - Bootstrap Sampling

# Bootstrap sampling creates repeated training views.
# This script shows replacement using tiny data.
# It supports bagging ideas without training models.

import numpy as np
import pandas as pd

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create a tiny patient style dataset.
data = pd.DataFrame({
    "patient_id": [101, 102, 103, 104, 105, 106],
    "risk_score": [0.20, 0.55, 0.40, 0.90, 0.35, 0.70],
})

# Validate the dataset before sampling.
assert data.shape == (6, 2)
assert data["patient_id"].is_unique

# Draw one bootstrap sample with replacement.
bootstrap_ids = rng.choice(
    data.index, size=len(data), replace=True
)

# Build the sampled dataset.
bootstrap_sample = data.iloc[bootstrap_ids].reset_index(drop=True)
counts = pd.Series(bootstrap_ids).value_counts().sort_index()

# Find repeated and omitted observations.
repeated = int((counts > 1).sum())
omitted = int(len(data) - counts.size)

# Show the original identifiers.
print("Original IDs:", data["patient_id"].tolist())

# Show the sampled identifiers.
print("Bootstrap IDs:", bootstrap_sample["patient_id"].tolist())

# Summarize replacement effects clearly.
print("Sample size matches original:", len(bootstrap_sample) == len(data))
print("Repeated original rows:", repeated)

# Report how many rows were omitted.
print("Omitted original rows:", omitted)
print("Unique rows used:", int(counts.size))

```
