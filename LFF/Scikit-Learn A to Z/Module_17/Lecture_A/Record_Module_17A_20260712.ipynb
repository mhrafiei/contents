## **2. Ensemble Randomization**

### **2.1. Bootstrap Sampling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_02_01.jpg?v=1783849814" width="250">



>* Bootstrap samples give each tree different data.
>* Different trees improve ensemble stability and accuracy.

>* Bootstrap sampling makes trees less similar.
>* Diverse trees average into steadier predictions.

>* Bootstrap forests resist noise and quirky cases.
>* Random resampling boosts diversity and generalization.



```
#@title Python Code - Bootstrap Sampling

# Bootstrap samples create different training views.
# Repeated rows increase tree diversity naturally.
# This script visualizes resampling overlap simply.

import numpy as np
import matplotlib.pyplot as plt

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Create tiny labeled training cases.
labels = np.array([
    "A", "B", "C", "D", "E", "F"
])

# Draw two bootstrap samples.
sample_size = labels.size
sample_one = rng.choice(labels, size=sample_size, replace=True)
sample_two = rng.choice(labels, size=sample_size, replace=True)

# Count appearances in each sample.
counts_one = np.array([
    np.sum(sample_one == label) for label in labels
])
counts_two = np.array([
    np.sum(sample_two == label) for label in labels
])

# Check sizes before plotting.
assert counts_one.size == labels.size
assert counts_two.size == labels.size

# Print a short interpretation.
missing_one = labels[counts_one == 0]
missing_two = labels[counts_two == 0]
print("Original cases:", ", ".join(labels))
print("Sample 1:", ", ".join(sample_one))
print("Sample 2:", ", ".join(sample_two))
print("Missing from sample 1:", ", ".join(missing_one) or "none")
print("Missing from sample 2:", ", ".join(missing_two) or "none")
print("Repeated rows change each tree's training view.")

# Plot how often each case appears.
x = np.arange(labels.size)
width = 0.38
plt.figure(figsize=(8, 4))

# Compare the two bootstrap samples.
plt.bar(x - width / 2, counts_one, width, label="Sample 1")
plt.bar(x + width / 2, counts_two, width, label="Sample 2")
plt.xticks(x, labels)
plt.ylabel("Times selected")

# Add a clear teaching title.
plt.title("Bootstrap sampling creates different datasets")
plt.legend()
plt.tight_layout()
plt.show()

```
