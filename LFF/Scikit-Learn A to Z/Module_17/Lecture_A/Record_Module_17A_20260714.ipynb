## **1. Forest Model Fitting**

### **1.1. Random Forest Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_01_01.jpg?v=1784009228" width="250">



>* Many trees vote on predicted classes
>* Ensembles improve stability for complex patterns

>* Train diverse trees using sampled data and features
>* Combine votes to capture complex customer patterns

>* Predict classes and probabilities for decisions
>* Evaluate beyond accuracy for imbalanced costs



```
#@title Python Code - Random Forest Classification

# Fit a random forest classifier.
# Compare predictions with held-out labels.
# Show accuracy and feature importance.

import sklearn
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
wine = load_wine()
features = wine.data
target = wine.target

# Validate the basic dataset shape.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Split data so testing uses unseen examples.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, stratify=target, random_state=42
)

# Fit many randomized trees as one forest.
forest = RandomForestClassifier(
    n_estimators=100, max_features="sqrt", random_state=42, n_jobs=-1
)
forest.fit(X_train, y_train)

# Evaluate predictions on the held-out test set.
predicted_labels = forest.predict(X_test)
accuracy = accuracy_score(y_test, predicted_labels)

# Identify the strongest impurity-based feature importance.
best_index = forest.feature_importances_.argmax()
best_feature = wine.feature_names[best_index]
best_importance = forest.feature_importances_[best_index]

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Test accuracy: {accuracy:.3f}")
print(f"Most important feature: {best_feature} ({best_importance:.3f})")

```

### **1.2. Random Forest Regression**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_01_02.jpg?v=1784009233" width="250">



>* Predicts numerical outcomes using many decision trees
>* Averaging trees improves stability and handles complexity

>* Trees learn diverse target-based splits
>* Averaged predictions handle complex, noisy patterns

>* Balance accuracy, stability, and generalization
>* Consider complexity, speed, and interpretability tradeoffs



```
#@title Python Code - Random Forest Regression

# Fit a random forest regression model.
# Average many trees for stable predictions.
# Compare predictions with held-out target values.

import sklearn
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Load a small packaged regression dataset.
diabetes = load_diabetes()
features = diabetes.data
target = diabetes.target

# Validate the basic supervised learning shapes.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target values.")

# Split data before fitting the forest.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42
)

# Fit many randomized trees and average their predictions.
forest = RandomForestRegressor(
    n_estimators=100, max_depth=5, random_state=42, n_jobs=-1
)
forest.fit(X_train, y_train)

# Evaluate predictions on data the model did not train on.
predictions = forest.predict(X_test)
mae = mean_absolute_error(y_test, predictions)

# Show a few predictions beside the true values.
print("scikit-learn version:", sklearn.__version__)
print("Test mean absolute error:", round(mae, 2))
print("First five true values:", [round(value, 1) for value in y_test[:5]])
print("First five predictions:", [round(value, 1) for value in predictions[:5]])

```

### **1.3. Extra Trees Fitting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_01_03.jpg?v=1784009232" width="250">



>* Randomized splits create diverse tree ensembles
>* Classify by votes, regress by averaging

>* Randomized trees learn diverse class patterns
>* Combined votes improve robust classification

>* Average randomized trees for continuous predictions
>* Tune, scale, and validate for generalization



```
#@title Python Code - Extra Trees Fitting

# Fit Extra Trees for a classification task.
# Randomized trees vote together for predictions.
# Accuracy and importances summarize the fitted model.

import sklearn
from sklearn.datasets import load_wine
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
wine = load_wine()
X = wine.data
y = wine.target

# Validate the basic feature and target shapes.
if X.shape[0] != y.shape[0]:
    raise ValueError("Features and target must have matching rows.")

# Split data so evaluation uses unseen examples.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Fit one Extra Trees ensemble with simple settings.
model = ExtraTreesClassifier(
    n_estimators=100, max_features="sqrt", random_state=42, n_jobs=-1
)
model.fit(X_train, y_train)

# Predict classes by aggregating votes from many trees.
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Show the strongest impurity-based feature importance.
best_index = int(model.feature_importances_.argmax())
best_feature = wine.feature_names[best_index]
best_importance = model.feature_importances_[best_index]

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Extra Trees test accuracy: {accuracy:.3f}")
print(f"Top impurity importance: {best_feature} = {best_importance:.3f}")

```

## **2. Randomization Settings**

### **2.1. Bootstrap Sampling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_02_01.jpg?v=1784009235" width="250">



>* Train trees on resampled data rows
>* Combine diverse trees for steadier predictions

>* Bootstrap diversity lowers tree variance
>* Combined predictions generalize more reliably

>* Bootstrap enables out-of-bag performance estimates
>* Choose based on diversity, data, and cost



```
#@title Python Code - Bootstrap Sampling

# This example shows bootstrap sampling in forests.
# Each tree receives a resampled training dataset.
# Out-of-bag rows estimate performance without validation data.

import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
data = load_breast_cancer()
X = data.data
y = data.target

# Validate the basic shape before modeling.
if X.shape[0] != y.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Split data so the test score remains independent.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.25, random_state=42
)

# Bootstrap enables out-of-bag scoring for random forests.
forest = RandomForestClassifier(
    n_estimators=100, bootstrap=True, oob_score=True, n_jobs=-1, random_state=42
)

# Fit one forest using many bootstrapped tree samples.
forest.fit(X_train, y_train)

# Compare built-in out-of-bag scoring with a held-out test score.
test_score = forest.score(X_test, y_test)
oob_score = forest.oob_score_

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Bootstrap sampling enabled: {forest.bootstrap}")
print(f"Out-of-bag score: {oob_score:.3f}")
print(f"Held-out test score: {test_score:.3f}")

```

### **2.2. Feature Subsampling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_02_02.jpg?v=1784009237" width="250">



>* Randomly limit features considered at each split
>* Encourage diverse trees and reduce predictor dominance

>* Reduces similarity between forest trees
>* Improves generalization through varied feature perspectives

>* Tune features to balance strength and diversity
>* Extra randomness improves speed and stability



```
#@title Python Code - Feature Subsampling

# This example compares feature subsampling settings.
# Forests become more diverse with fewer features.
# Test accuracy shows the practical tradeoff.

import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load a small packaged classification dataset.
data = load_breast_cancer()
X = data.data
y = data.target

# Validate the basic shape before modeling.
if X.shape[0] != y.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Split once so both forests face the same test data.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, stratify=y, random_state=42
)

# Train one forest that can inspect every feature.
all_features_forest = RandomForestClassifier(
    n_estimators=120, max_features=None, random_state=42, n_jobs=-1
)
all_features_forest.fit(X_train, y_train)

# Train one forest that samples features at each split.
subsampled_forest = RandomForestClassifier(
    n_estimators=120, max_features="sqrt", random_state=42, n_jobs=-1
)
subsampled_forest.fit(X_train, y_train)

# Compare accuracy and average tree agreement.
all_accuracy = accuracy_score(y_test, all_features_forest.predict(X_test))
sub_accuracy = accuracy_score(y_test, subsampled_forest.predict(X_test))

# Lower agreement means trees are making less identical predictions.
all_votes = [tree.predict(X_test) for tree in all_features_forest.estimators_]
sub_votes = [tree.predict(X_test) for tree in subsampled_forest.estimators_]

# Measure how often each tree matches the forest prediction.
all_prediction = all_features_forest.predict(X_test)
sub_prediction = subsampled_forest.predict(X_test)

all_agreement = sum((vote == all_prediction).mean() for vote in all_votes)
sub_agreement = sum((vote == sub_prediction).mean() for vote in sub_votes)

all_agreement = all_agreement / len(all_votes)
sub_agreement = sub_agreement / len(sub_votes)

print("scikit-learn version:", sklearn.__version__)
print("Dataset: breast cancer classification, features:", X.shape[1])
print("All features accuracy:", round(all_accuracy, 3))
print("sqrt feature subsampling accuracy:", round(sub_accuracy, 3))
print("All features average tree agreement:", round(all_agreement, 3))
print("sqrt subsampling average tree agreement:", round(sub_agreement, 3))

```

### **2.3. Out of Bag Scoring**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_02_03.jpg?v=1784009239" width="250">



>* Bootstrap leaves built-in test observations
>* Unused-tree predictions estimate validation performance

>* Uses limited data efficiently
>* Provides quick feedback for model tuning

>* Requires bootstrap sampling to be meaningful
>* Use careful validation for structured data



```
#@title Python Code - Out of Bag Scoring

# Demonstrate out-of-bag scoring in random forests.
# Bootstrap samples create built-in validation observations.
# Compare out-of-bag accuracy with test accuracy.

import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
data = load_breast_cancer()
X = data.data
y = data.target

# Check that features and labels match correctly.
if X.shape[0] != y.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Keep a final test set for comparison.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Enable bootstrap sampling and out-of-bag scoring.
forest = RandomForestClassifier(
    n_estimators=200, bootstrap=True, oob_score=True, n_jobs=-1, random_state=42
)

# Fit once; out-of-bag scoring is computed during training.
forest.fit(X_train, y_train)

# Compare internal validation with a held-out test set.
test_predictions = forest.predict(X_test)
test_accuracy = accuracy_score(y_test, test_predictions)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Training rows: {X_train.shape[0]}")
print(f"Out-of-bag accuracy: {forest.oob_score_:.3f}")
print(f"Held-out test accuracy: {test_accuracy:.3f}")
print("OOB scoring requires bootstrap=True in this forest.")

```

## **3. Forest Importance Inspection**

### **3.1. Parallel Forest Building**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_03_01.jpg?v=1784009223" width="250">



>* Independent trees train well in parallel
>* Parallelism speeds large, wide forest models

>* Parallel trees speed importance aggregation
>* Importance scores are not causal or unbiased

>* Parallelism enables faster importance stability checks
>* Importance still needs cautious validation



```
#@title Python Code - Parallel Forest Building

# This example compares parallel forest training settings.
# Feature importance is aggregated across many trees.
# Faster fitting does not change interpretation limits.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# A small synthetic dataset keeps the lesson reproducible.
features, target = make_classification(
    n_samples=1200,
    n_features=8,
    n_informative=3,
    n_redundant=1,
    random_state=42,
)

# Feature names make the importance scores easier to read.
feature_names = np.array([f"feature_{number}" for number in range(8)])
if features.shape[1] != len(feature_names):
    raise ValueError("Feature names must match the data columns.")

# The split is stratified because this is classification.
train_features, test_features, train_target, test_target = train_test_split(
    features,
    target,
    test_size=0.25,
    stratify=target,
    random_state=42,
)

# n_jobs=-1 lets scikit-learn build trees on available CPU cores.
forest = RandomForestClassifier(
    n_estimators=200,
    max_features="sqrt",
    n_jobs=-1,
    random_state=42,
)

# Fitting one forest is enough to inspect aggregated importance.
forest.fit(train_features, train_target)
accuracy = forest.score(test_features, test_target)
importances = forest.feature_importances_

# Sort features from most to least important.
order = np.argsort(importances)[::-1]
top_names = feature_names[order[:5]]
top_scores = importances[order[:5]]

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Test accuracy: {accuracy:.3f}")
print("Top impurity importances from the parallel forest:")
for name, score in zip(top_names, top_scores):
    print(f"{name}: {score:.3f}")

# The plot shows the same aggregated importance values.
fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(top_names, top_scores, color="steelblue")
ax.set_title("Top impurity-based feature importances")
ax.set_xlabel("Feature")
ax.set_ylabel("Importance score")
plt.show()

```

### **3.2. Impurity Importance**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_03_02.jpg?v=1784009229" width="250">



>* Splits reduce target uncertainty within tree nodes
>* Important features improve purity across many trees

>* Fast overview of model-used features
>* Predictive importance does not prove causation

>* Many split options can inflate importance
>* Validate with other inspection methods



```
#@title Python Code - Impurity Importance

# This example inspects forest impurity importance.
# Correlated features can share apparent importance.
# The plot ranks features after model training.

import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Create a small regression dataset with known informative features.
features, target = make_regression(
    n_samples=800,
    n_features=5,
    n_informative=2,
    noise=12.0,
    random_state=42,
)

# Add a correlated copy of the first informative feature.
correlated_copy = features[:, [0]] + 0.05 * features[:, [1]]
features = np.hstack([features, correlated_copy])
feature_names = np.array([
    "feature_0",
    "feature_1",
    "feature_2",
    "feature_3",
    "feature_4",
    "copy_of_0",
])

# Check that the feature matrix matches the feature names.
if features.shape[1] != len(feature_names):
    raise ValueError("Feature names must match the feature columns.")

# Split data before fitting the forest model.
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=42,
)

# Fit one small random forest for impurity importance.
forest = RandomForestRegressor(
    n_estimators=120,
    max_depth=6,
    random_state=42,
    n_jobs=-1,
)
forest.fit(X_train, y_train)

# Sort impurity importances from largest to smallest.
importances = forest.feature_importances_
order = np.argsort(importances)[::-1]
sorted_names = feature_names[order]
sorted_importances = importances[order]

# Print concise model and interpretation details.
score = r2_score(y_test, forest.predict(X_test))
print("scikit-learn version:", sklearn.__version__)
print("Test R2:", round(score, 3))
print("Top impurity importance:", sorted_names[0], round(sorted_importances[0], 3))
print("Correlated features can split importance between similar predictors.")

# Plot the impurity importance ranking.
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(sorted_names, sorted_importances, color="steelblue")
ax.set_title("Random Forest Impurity-Based Feature Importance")
ax.set_xlabel("Feature")
ax.set_ylabel("Mean impurity decrease")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

```

### **3.3. Missing Monotonic Constraints**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_17/Lecture_A/image_03_03.jpg?v=1784009225" width="250">



>* Forests may violate expected monotonic relationships
>* Importance scores miss directional consistency

>* Importance shows split usefulness, not trustworthy effects
>* High scores can hide irregular directions

>* Check prediction direction, not just importance
>* Use diagnostics or monotonic modeling safeguards



```
#@title Python Code - Missing Monotonic Constraints

# Forest importance can miss directional behavior.
# This example checks monotonic expectations visually.
# The plot reveals local prediction reversals.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import sklearn

# Synthetic data has a mostly increasing true relationship.
rng = np.random.default_rng(42)
debt_ratio = np.linspace(0, 1, 220)
noise = rng.normal(0, 0.08, size=debt_ratio.size)

# A small wave creates realistic sampling irregularity.
risk = 1.5 * debt_ratio + 0.25 * np.sin(18 * debt_ratio) + noise
X = debt_ratio.reshape(-1, 1)

# Validate the simple one-feature regression shape.
if X.shape[0] != risk.shape[0] or X.shape[1] != 1:
    raise ValueError("Expected one feature and matching target length.")

# Fit one forest model without monotonic constraints.
forest = RandomForestRegressor(
    n_estimators=120,
    random_state=42,
    min_samples_leaf=3,
)
forest.fit(X, risk)

# Inspect predictions across an ordered feature grid.
grid = np.linspace(0, 1, 160).reshape(-1, 1)
predicted_risk = forest.predict(grid)

# Count places where higher debt ratio lowers prediction.
drops = np.diff(predicted_risk) < -0.001
number_of_drops = int(np.sum(drops))
importance = float(forest.feature_importances_[0])

print("scikit-learn version:", sklearn.__version__)
print("Impurity importance for debt ratio:", round(importance, 3))
print("Local monotonicity violations found:", number_of_drops)

# Plot the learned relationship to reveal direction changes.
fig, ax = plt.subplots(figsize=(7, 4))
ax.scatter(debt_ratio, risk, s=14, alpha=0.35, label="training data")
ax.plot(grid.ravel(), predicted_risk, color="crimson", label="forest prediction")
ax.set_title("High importance does not guarantee monotonic behavior")
ax.set_xlabel("Debt-to-income ratio")
ax.set_ylabel("Predicted risk score")
ax.legend()
plt.show()

```
