## **1. Neighbor Search Methods**

### **1.1. Brute Force Search**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_13/Lecture_A/image_01_01.jpg?v=1783845080" width="250">



>* Compares every point to find neighbors.
>* Simple baseline, practical for small datasets.

>* Brute force slows as datasets grow.
>* Flexible for high dimensions and varied metrics.

>* Exact, flexible search for small complex datasets
>* Costs rise with many records and queries



```
#@title Python Code - Brute Force Search

# Brute force checks every candidate point.
# This script shows exact neighbor ranking.
# Small data keeps the idea clear.

import numpy as np

# Build a tiny two dimensional dataset.
points = np.array([
    [1.0, 1.0], [2.0, 1.5],
    [3.5, 4.0], [5.0, 4.5],
])

# Name each stored observation simply.
labels = np.array([
    "A", "B", "C", "D"
])

# Define one query point.
query = np.array([2.5, 2.0])

# Check shapes before distance calculations.
assert points.ndim == 2 and query.ndim == 1
assert points.shape[1] == query.shape[0]

# Compute Euclidean distances to all points.
diffs = points - query
sq_dists = np.sum(diffs ** 2, axis=1)
dists = np.sqrt(sq_dists)
order = np.argsort(dists)

# Choose the two nearest neighbors.
k = 2
nearest = order[:k]

# Report why brute force is exact.
print("Method: brute force over all stored points")
print(f"Query point: {query.tolist()}")
print(f"Compared candidates: {len(points)}")
print("All distances checked before choosing neighbors")

# Show the ranked nearest results.
for rank, idx in enumerate(nearest, start=1):
    print(
        f"Rank {rank}: {labels[idx]}, distance={dists[idx]:.2f}"
    )

# Show the full winner list.
print(f"Nearest labels: {labels[nearest].tolist()}")

```

### **1.3. Brute Force Search**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_13/Lecture_A/image_01_03.jpg?v=1783845081" width="250">



>* Compares queries against every stored point.
>* Exact results make it a baseline.

>* Brute force scales poorly with dataset size.
>* Works well for smaller, high-dimensional datasets.

>* Works with many custom distance measures.
>* Exact and simple, but costly at scale.



```
#@title Python Code - Brute Force Search

# Brute force checks every stored point.
# This example uses plain Python distances.
# It shows exact nearest neighbors clearly.

# Store tiny customer locations in meters.
points = [(0, 0), (2, 1), (4, 4), (7, 3)]
query = (3, 2)

# Compute Euclidean distance without special indexing.
distances = [
    (((x - query[0]) ** 2 + (y - query[1]) ** 2) ** 0.5, (x, y))
    for x, y in points
]

# Sort all candidates by distance.
distances.sort(key=lambda item: item[0])
nearest_two = distances[:2]

# Compare every point, then report winners.
print("Query point:", query)
print("Points checked:", len(points))
print("Nearest neighbor:", nearest_two[0][1], "distance", round(nearest_two[0][0], 2))
print("Second nearest:", nearest_two[1][1], "distance", round(nearest_two[1][0], 2))

# Show why brute force is flexible.
manhattan = [
    (abs(x - query[0]) + abs(y - query[1]), (x, y))
    for x, y in points
]

# Sort again using another metric.
manhattan.sort(key=lambda item: item[0])
print("Closest by Manhattan:", manhattan[0][1], "distance", manhattan[0][0])

```
