## **1. Maximum Margin Intuition**

### **1.1. Margin and Separation**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_01_01.jpg?v=1783844216" width="250">



>* Choose the boundary with widest margin.
>* Larger margins improve robustness to noise.

>* Wider margins reduce errors from uncertainty.
>* Separation improves reliability on new cases.

>* Closest opposing points determine the margin.
>* Maximum margin improves stability and generalization.



### **1.2. Support Vectors**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_01_02.jpg?v=1783844218" width="250">



>* Closest points constrain the decision boundary.
>* They determine the maximum margin width.

>* Borderline cases determine the classifier boundary.
>* SVMs rely mainly on influential points.

>* Distant points matter less than support vectors.
>* Hardest boundary cases shape the classifier.



### **1.3. Margin Classification Boundary**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_01_03.jpg?v=1783844221" width="250">



>* Boundary sits midway in widest gap
>* Larger buffers improve robustness to noise

>* Boundary depends on nearest training points.
>* It centers on hardest-to-separate cases.

>* Larger margins improve performance on new data.
>* Wide gaps create stable, resilient decisions.



## **2. SVC Variants**

### **2.1. Linear SVC Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_02_01.jpg?v=1783844211" width="250">



>* Linear SVC finds a flat separating boundary.
>* C controls margin strictness versus training errors.

>* Standardize features before fitting linear SVC.
>* Choose C using validation, not training.

>* Linear SVCs are fast, strong baselines.
>* They underfit curved patterns; kernels may help.



### **2.2. Kernel SVC Options**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_02_02.jpg?v=1783844213" width="250">



>* Kernel SVC models complex nonlinear class boundaries.
>* C balances training fit and generalization.

>* RBF kernel uses gamma for boundary reach.
>* Tune C and gamma with validation.

>* Polynomial adds flexibility; sigmoid is less reliable.
>* Standardize features and tune kernel settings carefully.



### **2.3. Kernel Choice**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_02_03.jpg?v=1783844215" width="250">



>* Linear suits sparse data; polynomial models interactions.
>* RBF handles complex nonlinear class boundaries.

>* Linear suits broad signals; RBF handles complexity.
>* Gamma and polynomial kernels need careful tuning.

>* Choose kernels by validation and task needs.
>* Prefer simpler models unless nonlinearity clearly helps.



## **3. SVM Tuning Choices**

### **3.1. Kernel Choice**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_03_01.jpg?v=1783844205" width="250">



>* Start linear for simple, high-dimensional problems.
>* Use nonlinear kernels for curved patterns.

>* Kernels differ in multiclass stability and flexibility.
>* Validate all classes, especially minority ones.

>* Probability needs and speed affect kernel choice.
>* Start simple; add flexibility only when needed.



### **3.2. Gamma Selection**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_03_02.jpg?v=1783844208" width="250">



>* Gamma controls how local boundaries become.
>* Too low underfits; too high overfits noise.

>* Gamma depends on task and local detail.
>* Validate gamma; high values can overfit.

>* High gamma can hurt multiclass imbalance handling.
>* Tune gamma for reliable probabilities and robustness.



### **3.3. Class Weighting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_12/Lecture_A/image_03_03.jpg?v=1783844206" width="250">



>* Weights help SVMs handle class imbalance.
>* They prioritize costly minority-class mistakes.

>* Weights trade recall against false alarms.
>* Judge weights using context-specific class metrics.

>* Weights help rare multiclass categories matter.
>* Validate carefully with kernels, regularization, and noise.
