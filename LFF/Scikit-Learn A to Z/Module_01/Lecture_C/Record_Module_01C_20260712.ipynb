## **2. Controlling Randomness**

### **2.1. Random Seed Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_01.jpg?v=1783829087" width="250">



>* Seeds start repeatable random number sequences.
>* They make splits and examples reproducible.

>* Random choices can change model results.
>* Fixed seeds make comparisons easier during development.

>* Seeds aid consistency, not universal truth.
>* Compare multiple seeds for dependable conclusions.



```
#@title Python Code - Random Seed Basics

# Random seeds make examples repeatable and easier.
# This script shows matching and changing randomness.
# Small outputs keep the idea easy.

import numpy as np

# Set one shared seed value.
seed = 7

# Generate numbers with the same seed.
np.random.seed(seed)
first_sample = np.random.randint(0, 10, size=5)

# Resetting repeats the same sequence.
np.random.seed(seed)
second_sample = np.random.randint(0, 10, size=5)

# A different seed changes results.
np.random.seed(seed + 1)
third_sample = np.random.randint(0, 10, size=5)

# Compare the three small samples.
print("Seed used:", seed)
print("First sample:", first_sample.tolist())
print("Second sample:", second_sample.tolist())
print("Third sample:", third_sample.tolist())

# Check whether repeated seeds match.
print("First equals second:", np.array_equal(first_sample, second_sample))
print("First equals third:", np.array_equal(first_sample, third_sample))

```

### **2.2. Random State Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_02.jpg?v=1783829091" width="250">



>* Random state makes random steps repeatable.
>* It enables fair comparisons and debugging.

>* Set random states across every workflow stage
>* Consistency reduces drift and debugging confusion

>* Fixed random states aid reproducibility, not truth.
>* Test multiple states for dependable conclusions.



```
#@title Python Code - Random State Basics

# Random seeds make repeated results easier.
# This example compares fixed and unfixed randomness.
# Small outputs keep the idea easy.

import random
import numpy as np

# Set one shared seed value.
seed = 7
random.seed(seed)
np.random.seed(seed)

# Generate repeatable values twice.
first_python = [random.randint(1, 9) for _ in range(5)]
first_numpy = np.random.randint(1, 10, size=5)

# Reset seeds and repeat.
random.seed(seed)
np.random.seed(seed)
second_python = [random.randint(1, 9) for _ in range(5)]
second_numpy = np.random.randint(1, 10, size=5)

# Generate values without resetting.
third_python = [random.randint(1, 9) for _ in range(5)]
third_numpy = np.random.randint(1, 10, size=5)

# Check that shapes are safe.
assert len(first_python) == 5 and first_numpy.shape == (5,)
assert len(second_python) == 5 and second_numpy.shape == (5,)

# Print short comparisons.
print("Seed value:", seed)
print("Python fixed run 1:", first_python)
print("Python fixed run 2:", second_python)
print("Python next run   :", third_python)
print("NumPy fixed run 1 :", first_numpy.tolist())
print("NumPy fixed run 2 :", second_numpy.tolist())
print("NumPy next run    :", third_numpy.tolist())
print("Python matches:", first_python == second_python)
print("NumPy matches:", np.array_equal(first_numpy, second_numpy))

```

### **2.3. Generated Example Seeds**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_03.jpg?v=1783829089" width="250">



>* Seeds determine the exact synthetic dataset.
>* Fixed seeds keep results consistent and comparable.

>* Generated data randomness affects task difficulty.
>* Fixed seeds enable fair model comparisons.

>* Choose and record seeds for reproducibility.
>* Test multiple seeds for stronger conclusions.



```
#@title Python Code - Generated Example Seeds

# Seeds keep generated examples reproducible.
# This script compares fixed and changed seeds.
# Small outputs make differences easy to inspect.

import numpy as np
import matplotlib.pyplot as plt

# Build one tiny synthetic dataset.
def make_points(seed, count=12):
    rng = np.random.default_rng(seed)
    x_values = rng.normal(loc=0.0, scale=1.0, size=count)
    y_values = 2.0 * x_values + rng.normal(scale=0.4, size=count)

    return np.column_stack((x_values, y_values))

# Generate datasets using chosen seeds.
points_a = make_points(seed=7)
points_b = make_points(seed=7)
points_c = make_points(seed=21)

# Check that shapes are safe.
assert points_a.shape == (12, 2)
assert points_b.shape == (12, 2)
assert points_c.shape == (12, 2)

# Compare repeated and changed seeds.
same_seed_match = np.allclose(points_a, points_b)
different_seed_match = np.allclose(points_a, points_c)
first_a = np.round(points_a[0], 2)
first_c = np.round(points_c[0], 2)

# Print a short interpretation.
print(f"Seed 7 repeated gives identical data: {same_seed_match}")
print(f"Seed 7 versus 21 identical: {different_seed_match}")
print(f"First point with seed 7: {first_a.tolist()}")
print(f"First point with seed 21: {first_c.tolist()}")

# Plot two generated examples side by side.
fig, axes = plt.subplots(ncols=2, figsize=(8, 3.5))
axes[0].scatter(points_a[:, 0], points_a[:, 1], color="tab:blue")
axes[0].set_title("Seed 7")
axes[0].set_xlabel("x value")

axes[0].set_ylabel("y value")
axes[1].scatter(points_c[:, 0], points_c[:, 1], color="tab:orange")
axes[1].set_title("Seed 21")
axes[1].set_xlabel("x value")

# Keep both panels comparable.
axes[1].set_ylabel("y value")
axes[0].set_xlim(-2.5, 2.5)
axes[1].set_xlim(-2.5, 2.5)
axes[0].set_ylim(-5.0, 5.0)

axes[1].set_ylim(-5.0, 5.0)
fig.suptitle("Same settings, different generated example seeds")
fig.tight_layout()
plt.show()

```

### **3.1. Data Loading**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_01.jpg?v=1783829064" width="250">



>* Start with a dedicated data loading section
>* Show data source and initial structure

>* Separate loading improves debugging and reproducibility.
>* Inspect shape, columns, and import problems early.

>* Clear data loading improves workflow communication.
>* It aids reuse, trust, and maintenance.



```
#@title Python Code - Data Loading

# This section isolates raw dataset loading.
# Small checks confirm source and structure.
# Clear starts improve notebook reproducibility.

# Import compact tools for loading examples.
import io
import numpy as np
import pandas as pd

# Fix randomness for reproducible sample values.
rng = np.random.default_rng(7)
text = io.StringIO(
    "city,temp_c,rain_mm\n"
    "Oslo,12.4,5\n"
    "Lima,18.1,0\n"
    "Tokyo,21.0,12\n"
)

# Load raw data in one dedicated step.
df = pd.read_csv(text)
df["sensor_id"] = rng.integers(100, 105, size=len(df))
summary = pd.DataFrame(
    {"column": df.columns, "dtype": df.dtypes.astype(str)}
)

# Inspect only the essentials, not everything.
print("Rows, columns:", df.shape)
print("Columns:", list(df.columns))
print("First row:", df.head(1).to_dict("records")[0])
print("Missing values:", int(df.isna().sum().sum()))

# Keep loading separate from later workflow stages.
print(summary.to_string(index=False))

```
