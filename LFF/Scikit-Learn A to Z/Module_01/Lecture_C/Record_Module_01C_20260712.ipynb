## **2. Reproducibility Basics**

### **2.1. Version Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_01.jpg?v=1783901440" width="250">



>* Record package versions for reproducible experiments
>* Different environments can change model results

>* Separate analysis errors from environment differences
>* Document versions across shared computing setups

>* Record package versions with each experiment
>* Clear environments make results easier to trust



```
#@title Python Code - Version Checks

# Version checks make notebooks easier to reproduce.
# Record important packages before running experiments.
# Keep outputs short for readable classroom notebooks.

# Install packages only if your environment lacks them.
# !pip install numpy pandas matplotlib seaborn.

# Import standard tools for environment inspection.
import platform
import random

# Import common data science packages.
import numpy as np
import pandas as pd

# Import visualization and plotting packages.
import matplotlib
import seaborn as sns

# Set one seed for repeatable random examples.
SEED_VALUE = 42
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)

# Store versions in a compact experiment record.
version_record = {
    "python": platform.python_version(),
    "numpy": np.__version__,
    "pandas": pd.__version__,
    "matplotlib": matplotlib.__version__,
    "seaborn": sns.__version__,
}

# Add reproducibility settings to the record.
version_record["random_seed"] = SEED_VALUE
version_record["notebook_role"] = "teaching example"

# Convert the record into a small table.
record_table = pd.DataFrame(
    list(version_record.items()),
    columns=["setting", "value"],
)

# Check the table size before printing.
row_count = len(record_table)
assert row_count <= 10, "Record table is unexpectedly long."

# Print a concise reproducibility summary.
print("Reproducibility record for this notebook:")
print(record_table.to_string(index=False))

# Create a tiny deterministic sample for verification.
sample_values = np.random.randint(1, 10, size=5)
print("Seed check sample:", sample_values.tolist())

```

### **2.2. Random Seed Control**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_02.jpg?v=1783901438" width="250">



>* Seeds make random steps repeatable
>* Controlled splits support fair model comparisons

>* Set seeds for every random step
>* Make model comparisons fair and auditable

>* Record seeds with versions and settings
>* Seeds help repeatability, not perfect reproduction



```
#@title Python Code - Random Seed Control

# Random seeds make examples repeatable.
# This script compares controlled random choices.
# Small outputs keep notebook results readable.

# Import built-in and numerical tools.
import random
import numpy as np

# Store one documented experiment seed.
experiment_seed = 42

# Create a tiny dataset of record identifiers.
record_ids = list(range(1, 11))

# Seed Python random before sampling records.
random.seed(experiment_seed)
python_sample_one = random.sample(record_ids, 4)

# Repeat the same seed before sampling again.
random.seed(experiment_seed)
python_sample_two = random.sample(record_ids, 4)

# Create NumPy generators with matching seeds.
rng_one = np.random.default_rng(experiment_seed)
rng_two = np.random.default_rng(experiment_seed)

# Generate two matching shuffled index orders.
order_one = rng_one.permutation(record_ids).tolist()
order_two = rng_two.permutation(record_ids).tolist()

# Create another generator with a different seed.
rng_three = np.random.default_rng(experiment_seed + 1)
order_three = rng_three.permutation(record_ids).tolist()

# Check that the tiny dataset is safe.
assert len(record_ids) == 10
assert len(python_sample_one) == 4

# Print concise reproducibility notes.
print(f"Documented seed: {experiment_seed}")
print(f"Python samples match: {python_sample_one == python_sample_two}")
print(f"Python sample: {python_sample_one}")
print(f"NumPy orders match: {order_one == order_two}")
print(f"Different seed changes order: {order_one != order_three}")
print(f"NumPy version: {np.__version__}")

```

## **3. Starter Workflow**

### **3.1. Load Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_01.jpg?v=1783901444" width="250">



>* Load data clearly and repeatably
>* Organize files to protect reproducibility

>* Document data sources and loading randomness
>* Use fixed seeds for reproducible rows

>* Check loaded data structure and target
>* Organize files for reproducible workflows



```
#@title Python Code - Load Dataset

# Load data with reproducible project habits.
# Keep paths clear and easy.
# Inspect small summaries before modeling.

import random
import numpy as np
import pandas as pd
from pathlib import Path

# Set one seed for repeatable sampling.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Create simple project folders safely.
project_dir = Path("starter_workflow_demo")
data_dir = project_dir / "data_raw"
data_dir.mkdir(parents=True, exist_ok=True)

# Build a tiny classroom dataset.
housing_data = pd.DataFrame({
    "home_id": [101, 102, 103, 104, 105, 106],
    "size_sqft": [850, 920, 1100, 1250, 1400, 1600],
    "distance_km": [2.1, 3.4, 1.8, 5.2, 4.0, 6.5],
    "price_usd": [210000, 235000, 275000, 310000, 355000, 390000],
})

# Save raw data in one known location.
csv_path = data_dir / "housing_sample.csv"
housing_data.to_csv(csv_path, index=False)

# Load data from the documented path.
loaded_data = pd.read_csv(csv_path)
expected_columns = {"home_id", "size_sqft", "distance_km", "price_usd"}

# Validate the loaded dataset structure.
row_count, column_count = loaded_data.shape
columns_ok = expected_columns.issubset(set(loaded_data.columns))
assert row_count > 0 and columns_ok

# Take a reproducible small sample.
sample_data = loaded_data.sample(n=3, random_state=SEED)
sample_ids = sample_data["home_id"].tolist()

# Print concise reproducibility notes.
print(f"Seed used: {SEED}")
print(f"Data path: {csv_path}")
print(f"Shape loaded: {row_count} rows, {column_count} columns")
print(f"Columns valid: {columns_ok}")
print(f"Sample home IDs: {sample_ids}")

```

### **3.2. Fit model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_02.jpg?v=1783901446" width="250">



>* Train models on prepared, consistent data
>* Set seeds for repeatable results

>* Separate fitting from other workflow steps
>* Improve debugging, accountability, and reproducibility

>* Use fixed seeds for fair model comparisons
>* Record fitting choices for reproducible projects



```
#@title Python Code - Fit model

# Fit models with repeatable starter workflow.
# Keep training separate from evaluation.
# Record seeds and project settings clearly.

import random
import numpy as np
import pandas as pd

# Store reproducibility settings in one place.
PROJECT_SETTINGS = {"seed": 42, "model": "simple_linear_regression"}

# Apply the same seed to common generators.
random.seed(PROJECT_SETTINGS["seed"])
np.random.seed(PROJECT_SETTINGS["seed"])

# Create a tiny housing-style training table.
housing = pd.DataFrame({"size_sqft": [650, 800, 950, 1100, 1250, 1400], "price_k": [180, 215, 250, 285, 320, 355]})

# Separate features and target before fitting.
X = housing[["size_sqft"]].to_numpy(dtype=float)
y = housing["price_k"].to_numpy(dtype=float)

# Validate the training data shape first.
assert X.shape[0] == y.shape[0]
assert X.shape[1] == 1

# Fit a simple line using NumPy.
slope, intercept = np.polyfit(X.ravel(), y, deg=1)

# Save fitted parameters in a tidy object.
fitted_model = {"slope": slope, "intercept": intercept}

# Predict one new home price consistently.
new_home_sqft = 1000.0
predicted_price = slope * new_home_sqft + intercept

# Print a compact reproducibility summary.
print(f"Seed used: {PROJECT_SETTINGS['seed']}")
print(f"Training rows: {len(housing)}")
print(f"Feature used: size_sqft")
print(f"Fitted slope: {slope:.3f} thousand dollars per sqft")
print(f"Prediction for 1000 sqft: ${predicted_price:.1f}k")

```

### **3.3. Score Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_03.jpg?v=1783901448" width="250">



>* Scores make model performance measurable
>* Treat scores as checkpoints, not final answers

>* Use fixed seeds for repeatable scores
>* Stable results support fair model comparisons

>* Organize scoring details for clear tracking
>* Use scores as reproducibility checkpoints



```
#@title Python Code - Score Model

# Score models with repeatable starter workflow habits.
# Use tiny data and documented random seeds.
# Keep scoring records organized and readable.

import random
import numpy as np
import pandas as pd

# Set one seed for repeatable splitting.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Create a tiny apartment rental dataset.
data = pd.DataFrame({
    "size_sqft": [500, 650, 800, 950, 1100, 1250, 1400, 1550],
    "rent_usd": [1200, 1450, 1700, 1950, 2200, 2450, 2700, 2950],
})

# Shuffle rows deterministically before splitting.
shuffled = data.sample(frac=1, random_state=SEED).reset_index(drop=True)
train = shuffled.iloc[:6].copy()
test = shuffled.iloc[6:].copy()

# Validate the split before modeling.
assert len(train) > 1 and len(test) > 0
assert set(train.columns) == {"size_sqft", "rent_usd"}

# Fit a simple line using NumPy.
slope, intercept = np.polyfit(train["size_sqft"], train["rent_usd"], 1)
predictions = slope * test["size_sqft"] + intercept
errors = test["rent_usd"] - predictions

# Score with mean absolute error.
mae = float(np.mean(np.abs(errors)))
baseline = float(np.mean(np.abs(test["rent_usd"] - train["rent_usd"].mean())))

# Store reproducibility details together.
record = {
    "seed": SEED,
    "train_rows": len(train),
    "test_rows": len(test),
    "score_mae_usd": round(mae, 2),
    "baseline_mae_usd": round(baseline, 2),
}

# Print a compact scoring checkpoint.
print("Starter workflow scoring checkpoint")
print(f"Random seed: {record['seed']}")
print(f"Train rows: {record['train_rows']}, test rows: {record['test_rows']}")
print(f"Model MAE: ${record['score_mae_usd']}")
print(f"Baseline MAE: ${record['baseline_mae_usd']}")
print("Lower MAE means closer rent predictions.")

```
