## **1. Display Settings**

### **1.1. Configure Display**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_01_01.jpg?v=1783975451" width="250">



>* Make estimator displays easier to understand
>* Clarify pipelines during early model comparisons

>* Choose displays that fit the context
>* Clear pipeline views reveal modeling decisions

>* Readable displays support sharing and reproducibility
>* Clear outputs reveal model and pipeline choices



```
#@title Python Code - Configure Display

# This example configures scikit-learn estimator displays.
# It compares compact text with diagram output.
# The printed settings confirm the active configuration.

import sklearn
from sklearn import config_context
from sklearn import get_config
from sklearn import set_config
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# A small packaged dataset keeps the example self-contained.
iris = load_iris()
feature_count = iris.data.shape[1]

# This validation checks the expected beginner-friendly dataset shape.
if feature_count != 4:
    raise ValueError("Expected the iris dataset to have four features.")

# A pipeline display is easier to inspect than a single estimator.
model_pipeline = Pipeline(
    steps=[
        ("scale", StandardScaler()),
        ("model", LogisticRegression(max_iter=200, random_state=42)),
    ]
)

# The global setting controls how estimators appear in notebooks.
set_config(display="diagram")
active_display = get_config()["display"]

print("scikit-learn version:", sklearn.__version__)
print("Global estimator display:", active_display)
print("Pipeline steps:", list(model_pipeline.named_steps.keys()))

# A temporary context can switch display style for one block.
with config_context(display="text"):
    temporary_display = get_config()["display"]
    print("Temporary display inside context:", temporary_display)

# After the context, the global diagram setting is restored.
restored_display = get_config()["display"]
print("Display after context:", restored_display)

```

### **1.2. HTML Estimator Diagrams**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_01_02.jpg?v=1783975453" width="250">



>* Turn estimators into visual workflow summaries
>* Inspect complex pipelines more easily

>* Visualize complex preprocessing workflows clearly
>* Spot missing, misplaced, or misapplied steps

>* Make model setup easy to revisit
>* Compare workflows, not just final algorithms



```
#@title Python Code - HTML Estimator Diagrams

# This example displays a pipeline as an HTML diagram.
# Scikit-learn configuration controls notebook estimator display.
# The output confirms the active display setting.

import sklearn
from sklearn import set_config
from sklearn.compose import ColumnTransformer
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Load a small packaged dataset for a quick demonstration.
iris = load_iris(as_frame=True)
features = iris.data
target = iris.target

# Validate the dataset shape before building the workflow.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target values.")

# Build a simple preprocessing step for all numeric columns.
numeric_columns = list(features.columns)
preprocessor = ColumnTransformer(
    transformers=[("scale_numeric", StandardScaler(), numeric_columns)]
)

# Combine preprocessing and modeling into one estimator object.
model_pipeline = Pipeline(
    steps=[("preprocess", preprocessor), ("classifier", LogisticRegression(max_iter=200))]
)

# Enable rich HTML diagrams for estimators in notebook displays.
set_config(display="diagram")

# Fit the pipeline so the displayed object represents a trained workflow.
model_pipeline.fit(features, target)

# Print concise reproducibility and configuration information.
print("scikit-learn version:", sklearn.__version__)
print("Estimator display mode: diagram")
print("In Colab, the final pipeline object appears as an HTML diagram.")

model_pipeline

```

## **2. Reproducibility Basics**

### **2.1. Version Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_01.jpg?v=1783975441" width="250">



>* Software versions affect analysis results
>* Record versions as experiment details

>* Versions explain changed results or warnings
>* Recorded environments support review and collaboration

>* Record Python and key package versions
>* Make environments visible for reproducible results



```
#@title Python Code - Version Checks

# This script records key package versions.
# Version checks support reproducible experiments.
# The output is a compact environment snapshot.

import numpy
import pandas
import matplotlib
import seaborn
import sklearn

# Store the versions in a small labeled dictionary.
version_snapshot = {
    "numpy": numpy.__version__,
    "pandas": pandas.__version__,
    "matplotlib": matplotlib.__version__,
    "seaborn": seaborn.__version__,
    "scikit-learn": sklearn.__version__,
}

# Validate that every recorded version is available.
missing_versions = []
for package_name, package_version in version_snapshot.items():
    if not package_version:
        missing_versions.append(package_name)

# Stop early if the environment snapshot is incomplete.
if missing_versions:
    raise RuntimeError("A package version could not be recorded.")

# Print a concise reproducibility note for a notebook.
print("Reproducibility snapshot for this notebook:")
for package_name in sorted(version_snapshot):
    package_version = version_snapshot[package_name]
    print(f"{package_name}: {package_version}")

# Confirm the habit this cell demonstrates.
print("Record these versions with your experiment results.")

```

### **2.2. Random Seed Control**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_02.jpg?v=1783975439" width="250">



>* Random steps can change experiment results
>* Fixed seeds make runs repeatable

>* Use fixed seeds wherever randomness appears
>* Keep comparisons fair across reruns and collaborators

>* Fixed seeds have reproducibility limits
>* Document seeds and environment details



```
#@title Python Code - Random Seed Control

# This example shows controlled random seed behavior.
# Fixed seeds make repeated splits reproducible.
# The output compares seeded and unseeded results.

import numpy as np
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic dataset shape before splitting.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Two unseeded splits may choose different test rows.
_, _, _, unseeded_test_a = train_test_split(
    features, target, test_size=0.3, stratify=target
)

_, _, _, unseeded_test_b = train_test_split(
    features, target, test_size=0.3, stratify=target
)

# Two seeded splits repeat the same random choices.
_, _, _, seeded_test_a = train_test_split(
    features, target, test_size=0.3, stratify=target, random_state=42
)

_, _, _, seeded_test_b = train_test_split(
    features, target, test_size=0.3, stratify=target, random_state=42
)

# NumPy randomness should also use a documented seed.
rng = np.random.default_rng(42)
example_numbers = rng.integers(0, 10, size=5)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Unseeded splits identical: {np.array_equal(unseeded_test_a, unseeded_test_b)}")
print(f"Seeded splits identical: {np.array_equal(seeded_test_a, seeded_test_b)}")
print(f"Documented seed value: 42")
print(f"Seeded NumPy sample: {example_numbers.tolist()}")

```

## **3. Starter Workflow**

### **3.1. Load Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_01.jpg?v=1783975445" width="250">



>* Load data clearly and repeatably
>* Track sources, versions, and random steps

>* Fix seeds for repeatable data samples
>* Compare results using shared starting conditions

>* Separate raw and processed data files
>* Use clear names and stable folders



```
#@title Python Code - Load Dataset

# Load data with reproducible starter workflow habits.
# Fixed seeds make early sampling repeatable.
# Clear labels document the dataset source.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Store one seed value for every random step.
random_seed = 42

# Load a packaged dataset with a documented source name.
iris = load_iris(as_frame=True)
features = iris.data
target = iris.target

# Validate the basic shape before continuing.
if len(features) != len(target):
    raise ValueError("Feature rows and target labels must match.")

# Create the same train and test rows every run.
train_features, test_features, train_target, test_target = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=random_seed,
    stratify=target,
)

# Repeat the split to prove the seed controls randomness.
repeat_train, repeat_test, repeat_y_train, repeat_y_test = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=random_seed,
    stratify=target,
)

same_test_rows = test_features.index.equals(repeat_test.index)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Dataset source: packaged sklearn iris dataset")
print(f"Project habit: raw data kept unchanged in memory")
print(f"Random seed used for split: {random_seed}")
print(f"Rows loaded: {len(features)}")
print(f"Train rows: {len(train_features)}, test rows: {len(test_features)}")
print(f"Repeated split gives same test rows: {same_test_rows}")

```

### **3.2. Fit Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_02.jpg?v=1783975447" width="250">



>* Fit models on prepared training data
>* Separate targets to prevent data leakage

>* Set seeds to control model randomness
>* Document seeds for reproducible comparisons

>* Organize files and notes for traceable fitting
>* Document runs to support trustworthy results



```
#@title Python Code - Fit Model

# Fit one model with reproducible starter workflow habits.
# Separate features, target, split, and seed clearly.
# Print version, settings, and repeatable evaluation results.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Store shared settings near the top for reproducibility.
seed = 42
project_note = "starter_workflow_fit_model"

# Load a small packaged dataset for a fast example.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic feature and target shapes before fitting.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Split once, using the same seed and stratified class balance.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=seed, stratify=target
)

# Fit one beginner-friendly model with the documented seed.
model = DecisionTreeClassifier(max_depth=3, random_state=seed)
model.fit(X_train, y_train)

# Evaluate the fitted model on data not used for fitting.
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Project note: {project_note}")
print(f"Documented random seed: {seed}")
print(f"Training rows used for fitting: {X_train.shape[0]}")
print(f"Test accuracy after one fit: {accuracy:.3f}")

```

### **3.3. Score Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_03.jpg?v=1783975449" width="250">



>* Score models on held-out evaluation data
>* Use fixed seeds for repeatable results

>* Keep scores linked to models and data
>* Record seeds and organize files clearly

>* Use the first score as a baseline
>* Keep seeds and structure consistent



```
#@title Python Code - Score Model

# Score a model with reproducible starter workflow habits.
# A fixed seed keeps evaluation splits consistent.
# The printed score becomes a clear baseline.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Keep one seed value near the top of the workflow.
seed = 42

# Load a small packaged classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic dataset shape before modeling.
if features.shape[0] != target.shape[0]:
    raise ValueError("Feature rows must match target labels.")

# Use the same seed for a repeatable train-test split.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=seed, stratify=target
)

# Fit one simple model with the same seed recorded.
model = DecisionTreeClassifier(max_depth=3, random_state=seed)
model.fit(X_train, y_train)

# Score the held-out test set as a baseline.
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Recorded random seed: {seed}")
print(f"Test rows scored: {len(y_test)}")
print(f"Baseline test accuracy: {accuracy:.3f}")

```
