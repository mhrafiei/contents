```
#@title Python Code - Version Checks

# Record versions before sharing notebooks.
# Reproducibility starts with environment notes.
# Small checks prevent confusing differences.

# Install packages only when Colab lacks them.
# !pip install numpy pandas matplotlib seaborn.

# Import standard tools for version reporting.
import platform, random, os

# Import common data science packages.
import numpy as np
import pandas as pd
import matplotlib

# Import seaborn for visualization version context.
import seaborn as sns

# Set one seed for repeatable examples.
SEED_VALUE = 42
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)

# Store versions in a compact dictionary.
version_notes = {
    "Python": platform.python_version(),
    "NumPy": np.__version__,
    "pandas": pd.__version__,
    "matplotlib": matplotlib.__version__,
    "seaborn": sns.__version__,
}

# Print a short reproducibility header.
print("Reproducibility version check")
print(f"Random seed: {SEED_VALUE}")

# Print each package version clearly.
for package_name, package_version in version_notes.items():
    print(f"{package_name}: {package_version}")

# Document the notebook execution context.
runtime_name = "Google Colab or compatible notebook"
print(f"Runtime note: {runtime_name}")

# Confirm the environment variable is optional.
api_key_present = os.getenv("OPENAI_API_KEY_LLF") is not None
print(f"Optional API key present: {api_key_present}")

# Create a tiny reproducibility record.
record = pd.DataFrame(list(version_notes.items()), columns=["item", "version"])

# Show only the compact record size.
print(f"Recorded rows: {len(record)}")

```

### **2.2. Random Seed Control**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_02_02.jpg?v=1783946148" width="250">



>* Randomness can change experiment results
>* Fixed seeds make reruns consistent

>* Use same seeds for fair model comparisons
>* Stable results improve transparency and debugging

>* Seeds help, but environments can still differ
>* Record shared seeds for reliable experiment tracking



```
#@title Python Code - Random Seed Control

# Random seeds make experiments repeatable.
# Fixed seeds support fair comparisons.
# Document seeds with package versions.

# Import small, standard scientific libraries.
import random, platform

# Import NumPy for controlled random numbers.
import numpy as np

# Record one shared seed value.
SEED_VALUE = 42

# Set Python and NumPy seeds.
random.seed(SEED_VALUE)
np.random.seed(SEED_VALUE)

# Create a tiny synthetic height dataset.
heights_cm = np.array([160, 165, 170, 175, 180, 185])

# Validate the dataset before sampling.
assert heights_cm.size >= 4, "Need at least four heights."

# Sample twice after resetting identical seeds.
np.random.seed(SEED_VALUE)
sample_one = np.random.choice(heights_cm, size=3, replace=False)

# Reset the seed before repeating.
np.random.seed(SEED_VALUE)
sample_two = np.random.choice(heights_cm, size=3, replace=False)

# Sample once without resetting seed.
sample_three = np.random.choice(heights_cm, size=3, replace=False)

# Check whether controlled samples match.
controlled_match = np.array_equal(sample_one, sample_two)

# Check whether uncontrolled sample differs.
uncontrolled_match = np.array_equal(sample_one, sample_three)

# Print concise reproducibility notes.
print(f"Python version: {platform.python_version()}")
print(f"NumPy version: {np.__version__}")
print(f"Documented seed: {SEED_VALUE}")
print(f"First seeded sample: {sample_one.tolist()} cm")
print(f"Repeated seeded sample: {sample_two.tolist()} cm")
print(f"Seeded samples match: {controlled_match}")
print(f"Next unreset sample matches: {uncontrolled_match}")

```

## **3. Starter Workflow**

### **3.1. Load Dataset**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_01.jpg?v=1783946150" width="250">



>* Load data from a reliable project folder
>* Use consistent paths for reproducible workflows

>* Check dataset shape, columns, and missing values
>* Verify targets and suspicious data before modeling

>* Reuse seeds for reproducible random operations
>* Organize data files for reliable workflows



```
#@title Python Code - Load Dataset

# Load data with reproducible project habits.
# Keep paths predictable and clearly named.
# Check small summaries before modeling.

# pandas is usually available in Colab.

# Import standard tools for organization.
from pathlib import Path
import random

# Import pandas for tabular loading.
import pandas as pd
import numpy as np

# Set one seed for reproducible sampling.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Create simple project folder paths.
project_dir = Path("starter_workflow_project")
raw_dir = project_dir / "data" / "raw"
raw_dir.mkdir(parents=True, exist_ok=True)

# Define a predictable raw data file.
data_path = raw_dir / "student_heights.csv"

# Build a tiny classroom dataset.
records = [
    {"student_id": 1, "height_cm": 168, "study_hours": 5, "passed": "yes"},
    {"student_id": 2, "height_cm": 155, "study_hours": 2, "passed": "no"},
    {"student_id": 3, "height_cm": 181, "study_hours": 6, "passed": "yes"},
    {"student_id": 4, "height_cm": 172, "study_hours": 3, "passed": "yes"},
]

# Save raw data only when missing.
if not data_path.exists():
    pd.DataFrame(records).to_csv(data_path, index=False)

# Load data from the predictable path.
data = pd.read_csv(data_path)

# Validate expected structure before analysis.
expected_columns = {"student_id", "height_cm", "study_hours", "passed"}
assert expected_columns.issubset(data.columns), "Missing expected columns."
assert len(data) > 0, "Dataset should not be empty."

# Create a reproducible small sample.
sample = data.sample(n=2, random_state=SEED)

# Print compact checks for learners.
print(f"Seed used: {SEED}")
print(f"Data path: {data_path}")
print(f"Shape: {data.shape[0]} rows, {data.shape[1]} columns")
print(f"Columns: {list(data.columns)}")
print(f"Missing values: {int(data.isna().sum().sum())}")
print(f"Sample IDs: {sample['student_id'].tolist()}")

```

### **3.2. Fit Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_02.jpg?v=1783946146" width="250">



>* Fit models using consistent data and settings
>* Set random seeds for repeatable results

>* Keep fitting steps simple and predictable
>* Use documented data, splits, and seeds

>* Fitted models store learned training information
>* Organized, seeded workflows improve reproducibility



```
#@title Python Code - Fit Model

# Fit models with reproducible starter workflow habits.
# Seeds and folders make experiments easier.
# This example avoids external machine learning libraries.

import os
import random
import numpy as np

# Use one seed across related workflow steps.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Keep simple project folders predictable.
project_root = "starter_workflow_demo"
data_folder = os.path.join(project_root, "data")
os.makedirs(data_folder, exist_ok=True)

# Create tiny housing data in square feet.
square_feet = np.array([650, 800, 950, 1100, 1250, 1400])
prices = np.array([180, 215, 250, 285, 320, 355])

# Validate matching feature and target lengths.
assert square_feet.shape[0] == prices.shape[0]
assert square_feet.shape[0] >= 4

# Make a reproducible train test split.
indices = np.arange(square_feet.shape[0])
np.random.shuffle(indices)
train_indices = indices[:4]
test_indices = indices[4:]

# Fit a simple line using training data.
train_x = square_feet[train_indices]
train_y = prices[train_indices]
slope, intercept = np.polyfit(train_x, train_y, deg=1)

# Evaluate predictions on held out examples.
test_x = square_feet[test_indices]
test_y = prices[test_indices]
predictions = slope * test_x + intercept
mae = np.mean(np.abs(predictions - test_y))

# Record compact reproducibility settings.
settings_path = os.path.join(project_root, "settings.txt")
settings_text = f"seed={SEED}, model=numpy_polyfit, split=4_train_2_test"
with open(settings_path, "w", encoding="utf-8") as file:
    file.write(settings_text)

# Print only concise workflow checkpoints.
print(f"NumPy version: {np.__version__}")
print(f"Project folder: {project_root}")
print(f"Random seed used: {SEED}")
print(f"Fitted line: price = {slope:.3f} * sqft + {intercept:.2f}")
print(f"Test MAE: {mae:.2f} thousand dollars")
print(f"Settings saved: {settings_path}")

```

### **3.3. Score Model**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_C/image_03_03.jpg?v=1783946156" width="250">



>* Evaluate using separate unseen test data
>* Keep scoring consistent and clearly documented

>* Set seeds to make scores reproducible
>* Compare models using the same split

>* Organize files, settings, seeds, and scores
>* Keep scoring reproducible, contextual, and honest



```
#@title Python Code - Score Model

# Score models with reproducible starter habits.
# Keep evaluation separate from training data.
# Record seeds, settings, and results clearly.

import random
import numpy as np
import pandas as pd

# Store one seed for the whole workflow.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Create a tiny housing style dataset.
home_size_sqft = np.array([650, 800, 950, 1100, 1250, 1400, 1550, 1700])
price_thousands = np.array([210, 240, 275, 310, 345, 380, 415, 450])

# Validate matching dataset lengths before splitting.
assert len(home_size_sqft) == len(price_thousands)
assert len(home_size_sqft) >= 6

# Shuffle indexes reproducibly using the seed.
indexes = np.arange(len(home_size_sqft))
np.random.shuffle(indexes)

# Keep test examples separate from training examples.
test_count = 2
train_indexes = indexes[:-test_count]
test_indexes = indexes[-test_count:]

# Build train and test arrays.
X_train = home_size_sqft[train_indexes]
y_train = price_thousands[train_indexes]
X_test = home_size_sqft[test_indexes]
y_test = price_thousands[test_indexes]

# Fit a simple line using NumPy only.
slope, intercept = np.polyfit(X_train, y_train, deg=1)
y_pred = slope * X_test + intercept

# Score with mean absolute error.
absolute_errors = np.abs(y_test - y_pred)
mae = float(np.mean(absolute_errors))

# Save compact experiment notes in one table.
summary = pd.DataFrame({
    "setting": ["seed", "train_rows", "test_rows", "metric"],
    "value": [SEED, len(X_train), len(X_test), "MAE"]
})

# Print only concise reproducibility information.
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Seed used: {SEED}")
print(f"Test MAE: {mae:.2f} thousand dollars")
print(summary.to_string(index=False))

```
