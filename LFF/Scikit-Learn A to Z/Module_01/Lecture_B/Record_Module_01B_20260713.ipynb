### **1.3. Supervised Versus Unsupervised**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_03.jpg?v=1783974664" width="250">



>* Supervised learning uses features and known targets
>* Models learn to predict unseen answers

>* Learns from features without target labels
>* Finds groups, patterns, or simpler representations

>* Labels indicate supervised prediction tasks
>* No labels means discovering hidden patterns



```
#@title Python Code - Supervised Versus Unsupervised

# This script contrasts supervised and unsupervised learning.
# It uses the same iris feature measurements.
# You will compare labels, fitting, and outputs.

import sklearn
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load a small packaged dataset with features and labels.
iris = load_iris()
features = iris.data[:, :2]
target = iris.target

# Validate that every sample has features and one target.
if features.shape[0] != target.shape[0]:
    raise ValueError("Each sample needs exactly one target label.")

# Supervised learning uses features and known target labels.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.3, stratify=target, random_state=42
)

supervised_model = LogisticRegression(max_iter=200, random_state=42)
supervised_model.fit(X_train, y_train)
supervised_predictions = supervised_model.predict(X_test)

# Unsupervised learning uses only features, not target labels.
unsupervised_model = KMeans(n_clusters=3, n_init=10, random_state=42)
unsupervised_model.fit(features)
cluster_ids = unsupervised_model.labels_

# Print concise results that highlight the training difference.
print("scikit-learn version:", sklearn.__version__)
print("Samples:", features.shape[0], "Features used:", features.shape[1])
print("Supervised fit call: model.fit(X_train, y_train)")
print("Supervised test accuracy:", round(accuracy_score(y_test, supervised_predictions), 2))
print("Unsupervised fit call: model.fit(X) only")
print("First 10 cluster ids:", cluster_ids[:10].tolist())

```

## **2. Task Types**

### **2.1. Classification Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_01.jpg?v=1783974667" width="250">



>* Assign examples to predefined category labels
>* Learn patterns separating known classes

>* Binary or multiclass categories
>* Classification chooses labels, not continuous values

>* Classification predicts discrete categories or decisions
>* Mistake costs depend on real-world context



```
#@title Python Code - Classification Tasks

# This example trains a simple classification model.
# Classification predicts categories instead of continuous numbers.
# The output shows labels and test accuracy.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load a small built-in multiclass classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate the basic samples, features, and target structure.
if features.shape[0] != target.shape[0]:
    raise ValueError("Each sample must have one target label.")

# Split labeled examples into training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=42, stratify=target
)

# Train one beginner-friendly classifier on labeled examples.
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

# Predict category labels for unseen test samples.
predicted_labels = classifier.predict(X_test)
accuracy = accuracy_score(y_test, predicted_labels)

# Convert numeric class codes into readable category names.
actual_name = iris.target_names[y_test[0]]
predicted_name = iris.target_names[predicted_labels[0]]

print(f"scikit-learn version: {sklearn.__version__}")
print(f"Samples: {features.shape[0]}, features per sample: {features.shape[1]}")
print(f"Target classes: {list(iris.target_names)}")
print(f"First test sample actual class: {actual_name}")
print(f"First test sample predicted class: {predicted_name}")
print(f"Test accuracy: {accuracy:.2f}")

```

### **2.2. Predicting Continuous Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_02.jpg?v=1783974669" width="250">



>* Regression predicts numerical values on a scale
>* Prediction size matters for practical decisions

>* Samples pair features with numerical targets
>* Models estimate useful values from past patterns

>* Regression predicts quantities, counts, rates, or scores
>* Evaluate errors by prediction distance and patterns



```
#@title Python Code - Predicting Continuous Values

# This example demonstrates a simple regression task.
# Regression predicts numbers instead of category labels.
# The output shows prediction errors have sizes.

import sklearn
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Load a small packaged dataset with a numerical target.
diabetes = load_diabetes()

# Use one feature so the idea stays easy to see.
feature_index = diabetes.feature_names.index("bmi")
X = diabetes.data[:, [feature_index]]
y = diabetes.target

# Check that samples and targets match before modeling.
if X.shape[0] != y.shape[0]:
    raise ValueError("Each sample must have one target value.")

# Split examples into training and testing data.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Fit one simple regression model on known examples.
model = LinearRegression()
model.fit(X_train, y_train)

# Predict numerical values for unseen test samples.
predictions = model.predict(X_test)
mean_error = mean_absolute_error(y_test, predictions)

# Compare a few true and predicted continuous values.
print(f"scikit-learn version: {sklearn.__version__}")
print("Task type: regression, because the target is numerical.")
print(f"Mean absolute error: {mean_error:.1f} target units")
print(f"Example true value: {y_test[0]:.1f}")
print(f"Example predicted value: {predictions[0]:.1f}")

```

### **2.3. Clustering Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_03.jpg?v=1783974672" width="250">



>* Finds natural groups without target labels
>* Clusters come from data similarity patterns

>* Find hidden patterns, not known labels
>* Explore meaningful groups in real examples

>* Choose relevant features and similarity measures
>* Interpret clusters with domain judgment



```
#@title Python Code - Clustering Tasks

# This example shows clustering without target labels.
# KMeans finds groups from feature similarity.
# The plot reveals inferred clusters visually.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import __version__ as sklearn_version
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Create unlabeled samples with two measurable features.
features, hidden_groups = make_blobs(
    n_samples=180,
    centers=3,
    cluster_std=0.75,
    random_state=42,
)

# Validate the feature table before clustering.
if features.shape != (180, 2):
    raise ValueError("Expected 180 samples with 2 features.")

# Fit KMeans using only features, not the hidden groups.
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(features)

# Count how many samples KMeans placed in each cluster.
cluster_counts = np.bincount(cluster_labels, minlength=3)
count_text = ", ".join(str(int(count)) for count in cluster_counts)

print(f"scikit-learn version: {sklearn_version}")
print("Input data: 180 samples, 2 features, no target column used.")
print(f"KMeans inferred cluster sizes: {count_text}")

# Plot the inferred clusters and their centers.
fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(
    features[:, 0],
    features[:, 1],
    c=cluster_labels,
    cmap="viridis",
    s=35,
)

# Mark the learned cluster centers for interpretation.
ax.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    c="red",
    marker="X",
    s=180,
    label="Cluster centers",
)

ax.set_title("Clustering Finds Groups Without Known Labels")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.legend()
plt.show()

```

## **3. Data Framing**

### **3.1. Samples and Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_01.jpg?v=1783974656" width="250">



>* Samples are individual cases models learn from
>* Features describe samples for modeling decisions

>* Rows are samples; columns are features
>* Keep feature meanings consistent across rows

>* Targets are outcomes models learn to predict
>* Use only realistic features to avoid leakage



```
#@title Python Code - Samples and Features

# Frame rows as samples for learning.
# Separate descriptive features from target labels.
# Show train and test evaluation splits.

import pandas as pd
from sklearn.model_selection import train_test_split

# Build a tiny housing dataset in memory.
homes = pd.DataFrame(
    {
        "area_sq_m": [55, 72, 80, 95, 110, 130, 150, 165],
        "bedrooms": [1, 2, 2, 3, 3, 4, 4, 5],
        "distance_km": [1.2, 3.5, 2.1, 5.0, 4.2, 6.8, 7.5, 8.1],
        "price_k": [210, 260, 300, 340, 390, 450, 520, 560],
    }
)

# Features describe each sample, while target is predicted.
feature_columns = ["area_sq_m", "bedrooms", "distance_km"]
target_column = "price_k"

# X is the feature matrix, and y is the target vector.
X = homes[feature_columns]
y = homes[target_column]

# Validate that every sample has one target value.
if len(X) != len(y):
    raise ValueError("Each sample must have exactly one target value.")

# Split samples before evaluation to keep test data separate.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Dataset framing for a supervised regression problem:")
print(f"Samples: {X.shape[0]}")
print(f"Features: {X.shape[1]} -> {list(X.columns)}")
print(f"Target: {target_column}")
print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
print("First two feature rows:")
print(X.head(2).to_string(index=False))

```

### **3.2. Train Validation Test**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_02.jpg?v=1783974653" width="250">



>* Split data for learning, tuning, and testing
>* Separate test data estimates real-world performance

>* Train, validate, then test like an exam
>* Keep test data separate to avoid overfitting

>* Match splits to data structure
>* Simulate real future model use



```
#@title Python Code - Train Validation Test

# This script frames data into honest evaluation splits.
# It demonstrates training, validation, and test roles.
# You will see split sizes and class balance.

import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load a small packaged classification dataset.
iris = load_iris()
features = iris.data
target = iris.target

# Validate that samples and targets match.
if features.shape[0] != target.shape[0]:
    raise ValueError("Each sample needs exactly one target value.")

# First hold back the final test set.
features_train_val, features_test, target_train_val, target_test = train_test_split(
    features, target, test_size=0.2, stratify=target, random_state=42
)

# Then split the remaining data into training and validation.
features_train, features_val, target_train, target_val = train_test_split(
    features_train_val, target_train_val, test_size=0.25,
    stratify=target_train_val, random_state=42
)

# Count samples in each split.
total_samples = len(target)
train_count = len(target_train)
val_count = len(target_val)
test_count = len(target_test)

# Show how the rows are framed for model development.
print(f"scikit-learn version: {sklearn.__version__}")
print(f"Total samples: {total_samples}, features per sample: {features.shape[1]}")
print(f"Training set: {train_count} samples for fitting the model")
print(f"Validation set: {val_count} samples for model choices")
print(f"Test set: {test_count} samples for final evaluation")

# Show that stratification preserves class balance.
train_balance = [round(100 * sum(target_train == c) / train_count, 1) for c in range(3)]
val_balance = [round(100 * sum(target_val == c) / val_count, 1) for c in range(3)]
test_balance = [round(100 * sum(target_test == c) / test_count, 1) for c in range(3)]

# Print compact class percentages for each split.
print(f"Class percentages in train: {train_balance}")
print(f"Class percentages in validation: {val_balance}")
print(f"Class percentages in test: {test_balance}")
print("Result: the test set stays untouched until the final exam.")

```
