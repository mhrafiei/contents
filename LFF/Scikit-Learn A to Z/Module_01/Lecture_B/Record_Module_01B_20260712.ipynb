### **1.3. Supervised Versus Unsupervised**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_03.jpg?v=1783900480" width="250">



>* Supervised learning trains with labeled examples
>* Models predict targets for new data

>* Learns from data without target labels
>* Finds patterns, groups, or simpler representations

>* Prediction goals use labeled supervised learning
>* Pattern discovery uses unlabeled unsupervised learning



```
#@title Python Code - Supervised Versus Unsupervised

# Compare learning families using tiny data.
# Supervised examples include known target answers.
# Unsupervised examples use features without targets.

import numpy as np

# Create a tiny feature matrix.
features = np.array([
    [650, 3, 1200],
    [900, 4, 1800],
    [500, 2, 900],
])

# Create supervised target values.
prices = np.array([250000, 410000, 190000])

# Create unsupervised customer features.
customers = np.array([
    [12, 80],
    [2, 15],
    [10, 75],
])

# Validate the supervised framing.
assert features.shape[0] == prices.shape[0]

# Validate the unsupervised framing.
assert customers.shape[1] == 2

# Summarize supervised learning inputs.
print("Supervised: fit with X features and y targets.")

# Show supervised dataset shape.
print(f"House X shape: {features.shape}, y shape: {prices.shape}")

# Summarize unsupervised learning inputs.
print("Unsupervised: fit with X features only.")

# Show unsupervised dataset shape.
print(f"Customer X shape: {customers.shape}, no y array.")

# Connect examples to problem families.
print("Classification predicts categories; regression predicts numbers.")

# Connect examples to exploratory families.
print("Clustering groups samples; reduction compresses features.")

# Frame a simple evaluation split.
train_rows = [0, 1]

# Keep one row as unseen data.
test_rows = [2]

# Report the split idea.
print(f"Train samples: {len(train_rows)}, test samples: {len(test_rows)}")

```

## **2. Task Types**

### **2.1. Classification Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_01.jpg?v=1783900475" width="250">



>* Assign examples to predefined category labels
>* Learn from labeled data to classify new examples

>* Classification chooses categories, not measurements
>* Classes may be binary, multiclass, or multilabel

>* Define meaningful, consistent class labels carefully
>* Consider fairness, error costs, and real-world use



```
#@title Python Code - Classification Tasks

# Classification predicts categories from example features.
# Labels are known during supervised training.
# This tiny example avoids external libraries.

# Store samples as simple feature dictionaries.
samples = [
    {"length_cm": 2.0, "width_cm": 1.0},
    {"length_cm": 2.5, "width_cm": 1.2},

    {"length_cm": 8.0, "width_cm": 3.0},
    {"length_cm": 9.0, "width_cm": 3.5},
]

# Store target class labels separately.
targets = ["small", "small", "large", "large"]

# Split examples into training and evaluation parts.
train_samples = samples[:3]
train_targets = targets[:3]

eval_samples = samples[3:]
eval_targets = targets[3:]

# Validate matching sample and target counts.
assert len(samples) == len(targets)
assert len(eval_samples) == len(eval_targets)

# Define a simple rule-based classifier.
def classify_item(item):
    total_size = item["length_cm"] + item["width_cm"]
    return "large" if total_size >= 10 else "small"

# Predict labels for evaluation samples.
predictions = [classify_item(item) for item in eval_samples]

# Count correct categorical predictions.
correct = sum(p == y for p, y in zip(predictions, eval_targets))
accuracy = correct / len(eval_targets)

# Print a compact classification summary.
print("Task type: classification")
print("Samples:", len(samples), "Features per sample:", 2)
print("Target labels:", sorted(set(targets)))
print("Evaluation prediction:", predictions[0])
print("Evaluation truth:", eval_targets[0])
print("Evaluation accuracy:", accuracy)

```

### **2.2. Predicting Continuous Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_02.jpg?v=1783900471" width="250">



>* Regression predicts numbers on a scale
>* Classification predicts fixed category labels

>* Regression forecasts quantities across many fields
>* Models learn known outcomes to predict new cases

>* Regression predicts meaningful numerical differences
>* Accuracy supports better real-world decisions



```
#@title Python Code - Predicting Continuous Values

# Regression predicts numbers on continuous scales.
# This example uses simple house data.
# We compare predictions with actual values.

import numpy as np
import matplotlib.pyplot as plt

# Store one feature, house size.
sizes_sqft = np.array([650, 800, 950, 1100, 1250, 1400])

# Store the continuous target, price.
prices_k = np.array([210, 245, 285, 330, 360, 410])

# Validate matching sample counts.
assert sizes_sqft.shape[0] == prices_k.shape[0]

# Fit a simple straight line.
slope, intercept = np.polyfit(sizes_sqft, prices_k, 1)

# Predict prices for new homes.
new_sizes = np.array([900, 1200, 1500])
predicted_prices = slope * new_sizes + intercept

# Print a compact regression summary.
print("Task type: regression")
print("Feature: house size in square feet")
print("Target: price in thousands of dollars")

# Show continuous predictions clearly.
for size, price in zip(new_sizes, predicted_prices):
    print(f"{size} sqft -> predicted price: ${price:.1f}k")

# Plot examples and fitted line.
plt.scatter(sizes_sqft, prices_k, label="known examples")
plt.plot(sizes_sqft, slope * sizes_sqft + intercept, label="regression line")

# Label the regression plot.
plt.xlabel("House size, square feet")
plt.ylabel("Price, thousands of dollars")
plt.title("Regression predicts continuous values")

# Display one compact visual result.
plt.legend()
plt.show()

```

### **2.3. Clustering Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_03.jpg?v=1783900473" width="250">



>* Finds natural groups without labels
>* Groups similar samples using feature patterns

>* Find hidden groups, not predefined outcomes
>* No labels or continuous predictions involved

>* Clusters need human interpretation
>* Useful for exploration, not final answers



```
#@title Python Code - Clustering Tasks

# Clustering discovers groups without target labels.
# This example uses simple customer features.
# We implement tiny clustering with NumPy.

import numpy as np
import matplotlib.pyplot as plt

# Set a seed for repeatable results.
rng = np.random.default_rng(7)

# Create three unlabeled customer behavior groups.
cluster_a = rng.normal([2, 8], [0.4, 0.5], (12, 2))
cluster_b = rng.normal([7, 3], [0.5, 0.4], (12, 2))

cluster_c = rng.normal([8, 8], [0.4, 0.4], (12, 2))
features = np.vstack([cluster_a, cluster_b, cluster_c])

# Validate the sample and feature layout.
assert features.shape == (36, 2)

# Choose starting centers from real samples.
centers = features[[0, 12, 24]].copy()

# Repeat assignment and center updates briefly.
for step in range(6):
    distances = ((features[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
    labels = distances.argmin(axis=1)
    centers = np.array([features[labels == k].mean(axis=0) for k in range(3)])

# Summarize discovered clusters, not known targets.
print("Task type: clustering, an unsupervised learning problem")
print("Samples:", features.shape[0], "Features per sample:", features.shape[1])
print("Target column provided? No")
print("Discovered cluster sizes:", np.bincount(labels).tolist())

# Plot the discovered customer groups.
plt.figure(figsize=(6, 4))
plt.scatter(features[:, 0], features[:, 1], c=labels, cmap="viridis", s=60)
plt.scatter(centers[:, 0], centers[:, 1], c="red", marker="X", s=160)
plt.xlabel("Monthly store visits")
plt.ylabel("Average basket value, tens of dollars")
plt.title("Clustering finds groups without labels")
plt.show()

```

## **3. Data Framing**

### **3.1. Samples and Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_01.jpg?v=1783900484" width="250">



>* Samples are individual cases models learn from
>* Features describe samples in consistent structure

>* Choose samples based on the model’s decision
>* Keep each row’s meaning consistent

>* Choose usable, relevant feature types
>* Avoid leakage, bias, and invalid conclusions



```
#@title Python Code - Samples and Features

# Frame rows as samples.
# Frame columns as features.
# Keep targets separate.

# Import pandas for tabular framing.
import pandas as pd

# Create a tiny housing dataset.
homes = pd.DataFrame(
    {
        "square_feet": [850, 1200, 1600, 2100],
        "bedrooms": [1, 2, 3, 4],

        "distance_km": [1.2, 3.5, 5.0, 8.4],
        "price_usd": [220000, 310000, 410000, 560000],
    }
)

# Separate features from the target.
feature_names = ["square_feet", "bedrooms", "distance_km"]
target_name = "price_usd"

# Build feature matrix and target vector.
X = homes[feature_names]
y = homes[target_name]

# Validate the framing before modeling.
assert len(X) == len(y)
assert X.shape[1] == len(feature_names)

# Make a simple evaluation split.
train_rows = [0, 1, 2]
test_rows = [3]

# Select training and testing samples.
X_train = X.iloc[train_rows]
X_test = X.iloc[test_rows]

# Select matching target values.
y_train = y.iloc[train_rows]
y_test = y.iloc[test_rows]

# Print a compact framing summary.
print(f"Samples: {len(homes)} rows")
print(f"Features: {list(X.columns)}")
print(f"Target: {target_name}")
print(f"X shape: {X.shape}")
print(f"y length: {len(y)}")
print(f"Train samples: {len(X_train)}")
print(f"Test samples: {len(X_test)}")
print("One sample features:", X_test.iloc[0].to_dict())

```

### **3.2. Evaluation Splits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_02.jpg?v=1783900491" width="250">



>* Train on some data, test on new data
>* Held-out tests reveal real generalization

>* Train, validate, and test separately
>* Protect final evaluation from tuning bias

>* Match splits to real-world use
>* Handle time, groups, and rare classes



```
#@title Python Code - Evaluation Splits

# Evaluation splits protect honest model assessment.
# We frame samples, features, and targets.
# Small synthetic data keeps output readable.

import random
import pandas as pd

# Use a deterministic seed for reproducible splitting.
random.seed(7)

# Create tiny supervised data with features and targets.
rows = []
for applicant_id in range(1, 13):
    income = 35 + applicant_id * 4
    debt = 18 + applicant_id % 5
    defaulted = int(debt > 20 and income < 70)
    rows.append([applicant_id, income, debt, defaulted])

# Store samples as a small labeled table.
columns = ["id", "income_k", "debt_k", "defaulted"]
data = pd.DataFrame(rows, columns=columns)

# Separate features from the supervised target.
features = data[["income_k", "debt_k"]]
target = data["defaulted"]

# Shuffle sample positions before assigning split names.
indices = list(data.index)
random.shuffle(indices)

# Use simple train, validation, and test sizes.
train_ids = indices[:7]
valid_ids = indices[7:9]
test_ids = indices[9:]

# Validate that every sample appears exactly once.
all_ids = train_ids + valid_ids + test_ids
assert len(set(all_ids)) == len(data)

# Attach split labels without changing feature values.
data["split"] = "unused"
data.loc[train_ids, "split"] = "train"
data.loc[valid_ids, "split"] = "validation"
data.loc[test_ids, "split"] = "test"

# Summarize the framed dataset without printing everything.
split_counts = data["split"].value_counts().sort_index()
feature_shape = features.shape
target_shape = target.shape

# Print concise teaching results for learners.
print("Samples:", len(data))
print("Feature matrix shape:", feature_shape)
print("Target vector shape:", target_shape)
print("Split counts:", split_counts.to_dict())
print("Test ids held back:", sorted(data.loc[test_ids, "id"].tolist()))

```
