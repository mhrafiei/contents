### **1.3. Supervised Versus Unsupervised**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_03.jpg?v=1783911113" width="250">



>* Supervised learning uses labeled target answers
>* Unsupervised learning finds patterns without answers

>* Supervised models predict known target outcomes
>* Unsupervised models find patterns without labels

>* Supervised predicts known outcomes from labeled examples
>* Unsupervised finds patterns before targets exist



```
#@title Python Code - Supervised Versus Unsupervised

# Compare learning problems using tiny examples.
# Supervised data includes known target answers.
# Unsupervised data has features without answers.

# Import small numerical tools for examples.
import numpy as np

# Create supervised house records with targets.
house_features = np.array([[900, 2], [1200, 3], [1500, 3], [1800, 4]])
house_prices = np.array([220000, 310000, 390000, 470000])

# Create unsupervised customer records without targets.
customer_features = np.array([[2, 20], [3, 25], [12, 180], [14, 210]])
customer_targets = None

# Split supervised samples into train and test.
train_features = house_features[:3]
test_features = house_features[3:]
train_targets = house_prices[:3]
test_targets = house_prices[3:]

# Validate tiny shapes before summarizing.
assert train_features.shape[0] == train_targets.shape[0]
assert test_features.shape[0] == test_targets.shape[0]
assert customer_features.shape[1] == 2

# Summarize the supervised framing.
print("Supervised: features plus target answers.")
print(f"House samples: {house_features.shape[0]}, features: {house_features.shape[1]}.")
print("Task type: regression, because price is numeric.")

# Summarize the evaluation split.
print(f"Train samples: {len(train_targets)}, test samples: {len(test_targets)}.")
print("Test target is hidden during future prediction.")

# Summarize the unsupervised framing.
print("Unsupervised: features only, no target answers.")
print(f"Customer samples: {customer_features.shape[0]}, features: {customer_features.shape[1]}.")
print("Task type: clustering or dimensionality reduction.")

# Show the decision rule clearly.
print("Known answer present? supervised; otherwise unsupervised.")

```

## **2. Task Types**

### **2.1. Classification Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_01.jpg?v=1783911106" width="250">



>* Assign examples to predefined class labels
>* Learn patterns from known labeled examples

>* Classification predicts categories, not measured quantities
>* Tasks may be binary or multiclass

>* Classification finds hidden category patterns
>* Test generalization on separate labeled data



```
#@title Python Code - Classification Tasks

# Classification predicts labels from features.
# Targets are categories, not measurements.
# This example uses tiny flower data.

# Import small built-in tools.
import numpy as np

# Store samples as feature rows.
features = np.array([
    [5.1, 3.5, 1.4, 0.2],
    [4.9, 3.0, 1.4, 0.2],
    [6.0, 2.9, 4.5, 1.5],
    [5.7, 2.8, 4.1, 1.3],

    [6.5, 3.0, 5.8, 2.2],
    [7.6, 3.0, 6.6, 2.1],
])

# Store known class labels.
targets = np.array([
    "setosa", "setosa", "versicolor",
    "versicolor", "virginica", "virginica",
])

# Validate matching sample counts.
assert features.shape[0] == targets.shape[0]

# Count samples and features.
sample_count, feature_count = features.shape

# Show dataset framing.
print(f"Samples: {sample_count}, features per sample: {feature_count}")

# Show possible classification outputs.
class_names = sorted(set(targets.tolist()))
print(f"Possible labels: {class_names}")

# Create a tiny evaluation split.
train_features = features[:4]
test_features = features[4:]
train_targets = targets[:4]
test_targets = targets[4:]

# Show split sizes.
print(f"Training samples: {len(train_targets)}")
print(f"Evaluation samples: {len(test_targets)}")

# Classify by nearest training example.
distances = np.linalg.norm(
    train_features - test_features[0], axis=1
)

# Choose the closest known label.
nearest_index = int(np.argmin(distances))
predicted_label = train_targets[nearest_index]
actual_label = test_targets[0]

# Print one classification decision.
print(f"Predicted label: {predicted_label}")
print(f"Actual label: {actual_label}")

# Emphasize the task type.
print("Task type: classification, because the target is a label")

```

### **2.2. Predicting Continuous Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_02.jpg?v=1783911104" width="250">



>* Regression predicts numerical values from labeled examples
>* Accuracy depends on closeness to true values

>* Regression predicts values that vary by degree
>* Examples use known numbers to estimate future outcomes

>* Target type separates regression from classification
>* Regression errors guide evaluation and decisions



```
#@title Python Code - Predicting Continuous Values

# Regression predicts numbers, not categories.
# Targets vary smoothly across numeric scales.
# Errors measure distance from true values.

import numpy as np
import matplotlib.pyplot as plt

# Create tiny house-size feature data.
sizes_sqft = np.array([650, 800, 950, 1100, 1300, 1500])
prices_k = np.array([180, 215, 250, 285, 330, 370])

# Validate matching sample and target counts.
assert sizes_sqft.shape[0] == prices_k.shape[0]
assert sizes_sqft.size >= 2 and prices_k.size >= 2

# Fit a simple line using NumPy.
slope, intercept = np.polyfit(sizes_sqft, prices_k, 1)
new_size = 1200
predicted_price = slope * new_size + intercept

# Show the regression framing clearly.
print("Task type: regression")
print("Feature example: house size in square feet")
print("Target example: selling price in thousands")
print(f"Prediction for {new_size} sqft: ${predicted_price:.1f}k")

# Plot samples and the fitted prediction line.
plt.figure(figsize=(6, 4))
plt.scatter(sizes_sqft, prices_k, label="known samples")
plt.plot(sizes_sqft, slope * sizes_sqft + intercept, label="fitted line")
plt.scatter([new_size], [predicted_price], marker="x", s=100, label="new prediction")
plt.xlabel("House size, square feet")
plt.ylabel("Price, thousands of dollars")
plt.title("Regression predicts a continuous value")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

### **2.3. Clustering Tasks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_02_03.jpg?v=1783911109" width="250">



>* Find groups without predefined labels
>* Analysts interpret patterns after clustering

>* Find hidden groups in unlabeled data
>* Explore patterns before making decisions

>* Interpret clusters carefully; features shape results
>* Evaluate clusters by usefulness and context



```
#@title Python Code - Clustering Tasks

# Clustering finds groups without target labels.
# This example uses tiny customer features.
# We implement simple clustering without sklearn.

import numpy as np
import matplotlib.pyplot as plt

# Create deterministic customer feature data.
rng = np.random.default_rng(seed=7)

# Build three visible customer groups.
cluster_a = rng.normal([2, 25], [0.4, 4], (12, 2))
cluster_b = rng.normal([7, 80], [0.5, 6], (12, 2))

cluster_c = rng.normal([5, 45], [0.4, 5], (12, 2))
customers = np.vstack([cluster_a, cluster_b, cluster_c])

# Validate the sample feature matrix.
assert customers.shape == (36, 2)

# Choose starting centers from real samples.
centers = customers[[0, 12, 24]].copy()

# Repeat assignment and center updates.
for step in range(6):
    distances = np.linalg.norm(customers[:, None] - centers[None, :], axis=2)
    labels = np.argmin(distances, axis=1)

    for group in range(3):
        members = customers[labels == group]
        if len(members) > 0:
            centers[group] = members.mean(axis=0)

# Count samples in each discovered group.
counts = np.bincount(labels, minlength=3)

# Print a compact task interpretation.
print("Task type: clustering, because no target labels are provided.")
print(f"Samples: {customers.shape[0]}, features per sample: {customers.shape[1]}.")
print("Features: visits per month and average spending in dollars.")
print(f"Discovered cluster sizes: {counts.tolist()}.")
print("Analysts interpret clusters after seeing the patterns.")

# Plot the discovered customer groups.
plt.figure(figsize=(6, 4))
plt.scatter(customers[:, 0], customers[:, 1], c=labels, cmap="viridis", s=60)
plt.scatter(centers[:, 0], centers[:, 1], c="red", marker="X", s=160)

# Label axes and title clearly.
plt.xlabel("Store visits per month")
plt.ylabel("Average spending in dollars")
plt.title("Clustering: hidden customer groups")
plt.grid(True, alpha=0.3)

# Display exactly one teaching plot.
plt.show()

```

## **3. Data Framing**

### **3.1. Samples and Features**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_01.jpg?v=1783911119" width="250">



>* Samples are individual cases for learning
>* Features describe each sample consistently

>* Choose relevant, reliable, timely features
>* Avoid future information causing data leakage

>* Represent information as structured feature types
>* Separate inputs from targets when present



```
#@title Python Code - Samples and Features

# Frame rows as samples.
# Frame columns as features.
# Separate targets when supervised.

# Import pandas for small tabular examples.
import pandas as pd

# Create a tiny housing dataset.
homes = pd.DataFrame(
    {
        "area_sqft": [850, 1200, 1600, 2100],
        "bedrooms": [1, 2, 3, 4],
        "distance_km": [2.1, 5.4, 8.0, 12.5],
        "price_usd": [240000, 310000, 420000, 560000],
    }
)

# Select feature columns for learning.
feature_columns = ["area_sqft", "bedrooms", "distance_km"]

# Build the feature table X.
X = homes[feature_columns]

# Build the supervised target y.
y = homes["price_usd"]

# Count samples and features.
sample_count, feature_count = X.shape

# Make a simple evaluation split.
train_X = X.iloc[:3]
test_X = X.iloc[3:]
train_y = y.iloc[:3]
test_y = y.iloc[3:]

# Validate the framing before printing.
assert len(train_X) == len(train_y)
assert len(test_X) == len(test_y)
assert feature_count == len(feature_columns)

# Print a compact framing summary.
print(f"Samples: {sample_count}")
print(f"Features: {feature_count}")
print(f"Feature names: {list(X.columns)}")
print(f"Target name: {y.name}")
print(f"Train samples: {len(train_X)}")
print(f"Test samples: {len(test_X)}")
print("First sample features:")
print(X.iloc[0].to_dict())

```

### **3.2. Train Validation Test**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_02.jpg?v=1783911123" width="250">



>* Split data into train, validation, and test
>* Hold out data to check generalization

>* Validation guides choices but can become biased
>* Untouched test data gives final performance check

>* Match splits to real prediction conditions
>* Preserve time, groups, and class balance



```
#@title Python Code - Train Validation Test

# This example frames samples, features, and targets.
# We create train, validation, and test splits.
# The split sizes stay small and readable.

import random

# Make the random shuffle reproducible.
random.seed(7)

# Create tiny house samples with features.
houses = [
    {"size_sqft": 900, "bedrooms": 2, "price_k": 260},
    {"size_sqft": 1100, "bedrooms": 2, "price_k": 310},
    {"size_sqft": 1300, "bedrooms": 3, "price_k": 360},
    {"size_sqft": 1500, "bedrooms": 3, "price_k": 410},

    {"size_sqft": 1700, "bedrooms": 3, "price_k": 470},
    {"size_sqft": 1900, "bedrooms": 4, "price_k": 530},
    {"size_sqft": 2100, "bedrooms": 4, "price_k": 590},
    {"size_sqft": 2300, "bedrooms": 4, "price_k": 650},

    {"size_sqft": 2500, "bedrooms": 5, "price_k": 720},
    {"size_sqft": 2700, "bedrooms": 5, "price_k": 790},
]

# Validate the dataset before splitting.
sample_count = len(houses)
assert sample_count == 10

# Shuffle row positions, not feature values.
indices = list(range(sample_count))
random.shuffle(indices)

# Choose simple sixty, twenty, twenty proportions.
train_end = int(sample_count * 0.60)
valid_end = int(sample_count * 0.80)

# Slice indices into three evaluation roles.
train_ids = indices[:train_end]
valid_ids = indices[train_end:valid_end]
test_ids = indices[valid_end:]

# Convert selected rows into feature matrices.
def make_features(row_ids):
    return [[houses[i]["size_sqft"], houses[i]["bedrooms"]] for i in row_ids]

# Convert selected rows into target vectors.
def make_targets(row_ids):
    return [houses[i]["price_k"] for i in row_ids]

# Build framed arrays for each split.
X_train = make_features(train_ids)
y_train = make_targets(train_ids)
X_valid = make_features(valid_ids)
y_valid = make_targets(valid_ids)

# Build the final untouched test split.
X_test = make_features(test_ids)
y_test = make_targets(test_ids)

# Confirm every sample appears exactly once.
all_ids = train_ids + valid_ids + test_ids
assert sorted(all_ids) == list(range(sample_count))

# Print compact teaching summaries only.
print("Samples:", sample_count)
print("Features per sample:", len(X_train[0]))
print("Target example: price_k")
print("Train rows:", len(X_train), "used to fit")

print("Validation rows:", len(X_valid), "used to choose")
print("Test rows:", len(X_test), "used at end")
print("First train features:", X_train[0])
print("First train target:", y_train[0])

```
