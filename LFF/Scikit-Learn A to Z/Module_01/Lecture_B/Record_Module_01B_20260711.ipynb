## **1. Problem Families**

### **1.1. Supervised Learning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_01.jpg?v=1783827936" width="250">



>* Learns from labeled input-output examples.
>* Examples include prices and spam detection.

>* Regression predicts numbers; classification predicts categories.
>* Known labels guide learning across data types.

>* Known past outcomes define supervised problems.
>* Useful decisions require accurate, unbiased labels.



```
#@title Python Code - Supervised Learning

# This script illustrates supervised learning basics.
# We use tiny labeled examples here.
# Outputs stay short and beginner friendly.

import numpy as np
import pandas as pd

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Build a tiny house dataset.
data = pd.DataFrame({
    "area_sqft": [900, 1100, 1400, 1600, 1900, 2200],
    "bedrooms": [2, 2, 3, 3, 4, 4],
    "price_usd": [180000, 210000, 265000, 300000, 360000, 410000],
})

# Separate samples, features, and target.
X = data[["area_sqft", "bedrooms"]].to_numpy(dtype=float)
y = data["price_usd"].to_numpy(dtype=float)

# Validate the tiny dataset shape.
assert X.shape[0] == y.shape[0], "Mismatched sample counts."
assert X.shape[1] == 2, "Expected two features."

# Create train, validation, test indices.
indices = rng.permutation(len(data))
train_idx, val_idx, test_idx = indices[:3], indices[3:5], indices[5:]

# Slice the dataset safely.
X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val = X[val_idx], y[val_idx]
X_test, y_test = X[test_idx], y[test_idx]

# Build a simple baseline predictor.
baseline_value = float(y_train.mean())
val_pred = np.full(shape=len(y_val), fill_value=baseline_value)

# Measure baseline error simply.
val_mae = float(np.mean(np.abs(y_val - val_pred)))
success_rule = "Lower validation MAE than baseline."

# Show why this is supervised.
print("Problem type: supervised regression")
print("Samples:", len(data), "houses")
print("Features:", ["area_sqft", "bedrooms"])
print("Target:", "price_usd")
print("Train/val/test sizes:", len(train_idx), len(val_idx), len(test_idx))
print("Baseline prediction:", round(baseline_value, 2))
print("Validation MAE:", round(val_mae, 2))
print("Success criterion:", success_rule)
print("Labels known during training:", bool(len(y_train) > 0))

```

### **1.2. Reduction Examples**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_02.jpg?v=1783827938" width="250">



>* Compress many features without predicting labels.
>* Creates simpler patterns from noisy retail data.

>* Images and text can have many features.
>* Reduction preserves patterns without needing labels.

>* Reduction supports visualization, noise removal, and modeling.
>* It compresses varied data into useful patterns.



```
#@title Python Code - Reduction Examples

# This example shows dimensionality reduction visually.
# We compress image features into two dimensions.
# Similar digits should stay near together.

# Import small tools for arrays and plotting.
import numpy as np
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Set a seed for reproducible choices.
np.random.seed(7)

# Load a tiny built in image dataset.
digits = load_digits()
X = digits.data
labels = digits.target

# Check that the dataset looks usable.
assert X.ndim == 2 and len(X) == len(labels)
assert X.shape[1] == 64 and len(X) > 100

# Reduce sixty four pixel features to two.
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Print a short interpretation summary.
print("Original feature count:", X.shape[1])
print("Reduced feature count:", X_reduced.shape[1])
print("Samples:", X.shape[0])
print("No target is predicted during reduction.")

# Plot the compact representation of images.
plt.figure(figsize=(7, 5))
scatter = plt.scatter(
    X_reduced[:, 0], X_reduced[:, 1],
    c=labels, s=18, cmap="tab10", alpha=0.8
)

# Add labels that explain the picture.
plt.title("Digits compressed from 64 features to 2")
plt.xlabel("Reduced feature 1")
plt.ylabel("Reduced feature 2")
plt.colorbar(scatter, label="Digit label")

# Show the final beginner friendly result.
plt.tight_layout()
plt.show()

```

### **1.3. Supervised Learning**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_01_03.jpg?v=1783827942" width="250">



>* Learns from labeled input-output examples.
>* Examples include spam detection and price prediction.

>* Supervised learning includes classification and regression.
>* It learns from labeled input-output examples.

>* Labeled data defines supervised learning tasks.
>* Tables, text, and images can have targets.



```
#@title Python Code - Supervised Learning

# Supervised learning uses known answers.
# Inputs map toward target outputs.
# This example shows labels clearly.

import numpy as np
import pandas as pd

# Set a deterministic random seed.
rng = np.random.default_rng(7)

# Build a tiny labeled dataset.
study_hours = np.array([1, 2, 3, 4, 5, 6])
passed_exam = np.array([0, 0, 0, 1, 1, 1])

# Create a simple table view.
df = pd.DataFrame({
    "study_hours": study_hours,
    "passed_exam": passed_exam,
})

# Separate features from target labels.
X = df[["study_hours"]]
y = df["passed_exam"]

# Check the dataset shape safely.
assert X.shape == (6, 1)
assert y.shape == (6,)

# Make a tiny train-test split.
train_idx = np.array([0, 1, 3, 4])
test_idx = np.array([2, 5])

# Compute class averages from training.
mean_fail = X.iloc[y.iloc[train_idx].to_numpy() == 0].mean().iloc[0]
mean_pass = X.iloc[y.iloc[train_idx].to_numpy() == 1].mean().iloc[0]

# Predict by nearest class average.
test_hours = X.iloc[test_idx, 0].to_numpy()
pred = np.where(
    np.abs(test_hours - mean_fail) < np.abs(test_hours - mean_pass),
    0,
    1,
)

# Compare predictions with known labels.
true_labels = y.iloc[test_idx].to_numpy()
accuracy = (pred == true_labels).mean()

# Print a short teaching summary.
print("Problem type: supervised classification")
print("Samples:", len(df), "Features:", X.shape[1])
print("Feature column:", list(X.columns)[0])
print("Target label column:", y.name)
print("Known labels:", true_labels.tolist())
print("Predicted labels:", pred.tolist())
print("Baseline idea: guess the majority class")
print("Success criterion: test accuracy")
print("Test accuracy:", round(float(accuracy), 2))

```

## **3. Evaluation Basics**

### **3.1. Train Validation Test**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_01.jpg?v=1783827925" width="250">



>* Train learns; validation guides model choices.
>* Test stays untouched for fair generalization check.

>* Validation checks for overfitting during development.
>* Test set estimates real-world performance.

>* Split data to match real-world use.
>* Prevent leakage; keep test set untouched.



```
#@title Python Code - Train Validation Test

# This script explains train validation test splits.
# It uses tiny built in Python data.
# Results stay short and beginner friendly.

# Create a tiny labeled housing dataset.
rows = [
    {"size_sqft": 900, "price_band": "low"},
    {"size_sqft": 1100, "price_band": "low"},
    {"size_sqft": 1300, "price_band": "medium"},
    {"size_sqft": 1500, "price_band": "medium"},
    {"size_sqft": 1700, "price_band": "medium"},
    {"size_sqft": 1900, "price_band": "high"},
    {"size_sqft": 2100, "price_band": "high"},
    {"size_sqft": 2300, "price_band": "high"},
    {"size_sqft": 1250, "price_band": "low"},
    {"size_sqft": 1600, "price_band": "medium"},
    {"size_sqft": 2000, "price_band": "high"},
    {"size_sqft": 1450, "price_band": "medium"},
]

# Shuffle deterministically for reproducible splits.
import random
random.seed(7)
random.shuffle(rows)

# Split into train validation and test.
train = rows[:7]
validation = rows[7:10]
test = rows[10:]

# Check split sizes before evaluation.
assert len(train) == 7 and len(validation) == 3
assert len(test) == 2 and len(rows) == 12

# Build a simple baseline from training labels.
train_labels = [row["price_band"] for row in train]
baseline_label = max(set(train_labels), key=train_labels.count)

# Score a constant prediction baseline.
def accuracy(data, predicted_label):
    correct = sum(
        row["price_band"] == predicted_label
        for row in data
    )
    return correct / len(data)

# Define a simple rule based model.
def rule_predict(size_sqft):
    if size_sqft < 1400:
        return "low"
    if size_sqft < 1850:
        return "medium"
    return "high"

# Score the rule based model.
def rule_accuracy(data):
    correct = sum(
        rule_predict(row["size_sqft"]) == row["price_band"]
        for row in data
    )
    return correct / len(data)

# Set a clear success criterion.
success_goal = 0.80
baseline_val = accuracy(validation, baseline_label)
rule_val = rule_accuracy(validation)
rule_test = rule_accuracy(test)

# Print a compact task summary.
print("Task: predict house price band from size.")
print("Feature: size_sqft | Target: price_band")
print(f"Split sizes -> train {len(train)}, val {len(validation)}, test {len(test)}")
print(f"Baseline label from train: {baseline_label}")
print(f"Baseline validation accuracy: {baseline_val:.2f}")
print(f"Rule validation accuracy: {rule_val:.2f}")
print(f"Success criterion: test accuracy >= {success_goal:.2f}")
print(f"Final test accuracy: {rule_test:.2f}")
print(f"Meets success criterion: {rule_test >= success_goal}")

```

### **3.2. Validation Split**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_02.jpg?v=1783827923" width="250">



>* Validation checks models during development.
>* It helps detect overfitting and compare options.

>* Match validation data to real deployment.
>* Keep time and groups split properly.

>* Keep validation fixed; avoid over-tuning decisions.
>* Use it to compare trade-offs fairly.



```
#@title Python Code - Validation Split

# Validation splits guide model choices safely.
# This example compares training and validation accuracy.
# Small data keeps the lesson fast.

import numpy as np
import pandas as pd

# Set a seed for repeatable results.
rng = np.random.default_rng(7)
# Create simple student support data.
study_hours = rng.normal(5.0, 1.5, 60)
attendance = rng.normal(85.0, 8.0, 60)

# Build labels from a hidden rule.
needs_help = (
    (study_hours < 4.5) |
    (attendance < 80.0)
).astype(int)

# Store the tiny dataset neatly.
data = pd.DataFrame({
    "study_hours": study_hours,
    "attendance": attendance,
    "needs_help": needs_help,
})

# Shuffle rows before splitting randomly.
data = data.sample(frac=1.0, random_state=7)
# Create train, validation, and test parts.
train_end = 36
valid_end = 48

# Slice the three dataset parts.
train = data.iloc[:train_end]
valid = data.iloc[train_end:valid_end]
test = data.iloc[valid_end:]

# Check split sizes before scoring.
assert len(train) and len(valid) and len(test)
# Build a simple baseline predictor.
majority_label = int(train["needs_help"].mode().iat[0])

# Predict one label for every row.
train_pred = np.full(len(train), majority_label)
valid_pred = np.full(len(valid), majority_label)
test_pred = np.full(len(test), majority_label)

# Compute accuracy for each split.
train_acc = (train_pred == train["needs_help"]).mean()
valid_acc = (valid_pred == valid["needs_help"]).mean()
test_acc = (test_pred == test["needs_help"]).mean()

# State the learning task clearly.
print("Task: predict student support needs")
print("Target column: needs_help")
print("Success criterion: highest validation accuracy")
print(f"Split sizes: {len(train)}, {len(valid)}, {len(test)}")

# Show why validation matters.
print(f"Baseline label from training: {majority_label}")
print(f"Training accuracy: {train_acc:.2f}")
print(f"Validation accuracy: {valid_acc:.2f}")
print(f"Test accuracy: {test_acc:.2f}")

```

### **3.3. Baseline Choice**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Scikit-Learn A to Z/Module_01/Lecture_B/image_03_03.jpg?v=1783827927" width="250">



>* Baselines give simple performance reference points.
>* Complex models should beat them meaningfully.

>* Baselines vary by task and data.
>* Keep them simple, relevant, and reproducible.

>* Baselines make success meaningful and practical.
>* They support honest evaluation and better decisions.



```
#@title Python Code - Baseline Choice

# Baselines give simple comparison points.
# This script frames one tiny task.
# We compare naive house-price guesses.

# Import small built-in tools.
import statistics as stats

# Create a tiny tabular dataset.
houses = [
    {"size_sqft": 900, "price": 180000},
    {"size_sqft": 1100, "price": 220000},
    {"size_sqft": 1300, "price": 260000},
    {"size_sqft": 1500, "price": 300000},
]

# Split into train, validation, test.
train = houses[:2]
valid = houses[2:3]
test = houses[3:]

# Extract target values safely.
train_prices = [row["price"] for row in train]
valid_prices = [row["price"] for row in valid]
test_prices = [row["price"] for row in test]

# Build two simple baselines.
mean_guess = round(stats.mean(train_prices))
last_guess = train_prices[-1]

# Define a tiny error function.
def mae(actual, predicted):
    return sum(abs(a - p) for a, p in zip(actual, predicted)) / len(actual)

# Score both baselines on validation.
valid_mean_preds = [mean_guess for _ in valid_prices]
valid_last_preds = [last_guess for _ in valid_prices]
mean_valid_mae = mae(valid_prices, valid_mean_preds)
last_valid_mae = mae(valid_prices, valid_last_preds)

# Choose the better validation baseline.
best_name = "mean price" if mean_valid_mae <= last_valid_mae else "last seen price"
best_guess = mean_guess if mean_valid_mae <= last_valid_mae else last_guess

# Test the chosen baseline once.
test_preds = [best_guess for _ in test_prices]
test_mae = mae(test_prices, test_preds)

# Print the task framing clearly.
print("Task: predict house price from size.")
print("Sample count:", len(houses))
print("Feature: size_sqft | Target: price")
print("Split sizes:", len(train), len(valid), len(test))
print("Baseline A guess:", mean_guess)
print("Baseline B guess:", last_guess)
print("Validation MAE A:", round(mean_valid_mae, 1))
print("Validation MAE B:", round(last_valid_mae, 1))
print("Chosen baseline:", best_name)
print("Success criterion: beat test MAE of", round(test_mae, 1))

```
