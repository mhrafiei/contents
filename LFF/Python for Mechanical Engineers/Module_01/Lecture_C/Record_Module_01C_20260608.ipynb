## **1. Core Data Types**

### **1.1. Floating Point Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_01_01.jpg?v=1780923403" width="250">



>* Floats represent decimal engineering measurements
>* Numeric floats enable calculations, unlike text

>* Floats handle very small and large decimals
>* Binary storage can cause tiny rounding errors

>* Use floats for continuous decimal quantities
>* Round results to meaningful engineering precision



```
#@title Python Code - Floating Point Values

# Floats represent measurements with decimal values.
# Engineering calculations often need fractional quantities.
# Formatting helps report realistic engineering precision.

# Store measurements as floating point values.
beam_length_m = 2.75
voltage_v = 3.3
pressure_kpa = 101.325

# Show each variable type clearly.
print("beam_length_m type:", type(beam_length_m).__name__)
print("voltage_v type:", type(voltage_v).__name__)
print("pressure_kpa type:", type(pressure_kpa).__name__)

# Convert an imperial measurement to metric.
bolt_diameter_in = 0.375
bolt_diameter_mm = bolt_diameter_in * 25.4

# Division often creates floating point results.
load_n = 1250.0
support_count = 3
load_per_support_n = load_n / support_count

# Small rounding effects can appear internally.
small_sum = 0.1 + 0.2

# Format results for engineering reporting.
print(f"Bolt diameter: {bolt_diameter_mm:.2f} mm")
print(f"Load per support: {load_per_support_n:.1f} N")
print(f"Raw floating result: {small_sum}")
print(f"Reported result: {small_sum:.1f}")

```

### **1.2. Integer Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_01_02.jpg?v=1780923405" width="250">



>* Integers represent exact whole-number counts
>* Use them for discrete engineering quantities

>* Integers index, count, and control calculations
>* Whole-number types clarify counts versus measurements

>* Same-looking numbers can have different types
>* Choose integers for counts, not labels



```
#@title Python Code - Integer Values

# Integers represent counted engineering quantities.
# This example compares related types.
# Counts differ from labels and measurements.

# Store counted components as integers.
bolt_count = 12
sensor_count = 4

# Store measured quantities as floats.
beam_length_m = 3.0
hole_diameter_in = 0.5

# Store identifier text as strings.
pump_id = "12"
assembly_name = "Bracket A"

# Store yes-or-no states as booleans.
inspection_passed = True
report_ready = False

# Print values with their Python types.
print("bolt_count:", bolt_count, type(bolt_count).__name__)
print("beam_length_m:", beam_length_m, type(beam_length_m).__name__)
print("pump_id:", pump_id, type(pump_id).__name__)

# Integers work naturally in count arithmetic.
total_fasteners = bolt_count + sensor_count
print("total counted items:", total_fasteners)

# Strings that look numeric are labels.
label_result = pump_id + "A"
print("pump label extension:", label_result)

# Booleans describe logical engineering states.
print("inspection_passed:", inspection_passed, type(inspection_passed).__name__)

# Format an integer count clearly.
summary = f"Assembly {assembly_name} uses {bolt_count} bolts."
print(summary)

```

### **1.3. String Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_01_03.jpg?v=1780923407" width="250">



>* Strings store text, not calculable quantities
>* Use strings for labels and context

>* Strings preserve exact wording and formatting
>* Strings label data for clearer interpretation

>* Numeric-looking strings need conversion before calculations
>* Use strings for labels, numbers for math



```
#@title Python Code - String Values

# Strings store engineering text labels.
# Numeric looking text needs conversion.
# Clear reports combine text and numbers.

# Store identifiers and units as strings.
sensor_tag = "PUMP-02 pressure sensor"
unit_label = "kPa"

# Store measurements as numeric values.
pressure_value = 248.6
limit_value = 300.0

# Store numeric characters as text.
pressure_text = "248.6"
combined_text = pressure_text + unit_label

# Convert text before calculations.
converted_pressure = float(pressure_text)
safety_margin = limit_value - converted_pressure

# Check each value type explicitly.
print("sensor_tag type:", type(sensor_tag).__name__)
print("pressure_text type:", type(pressure_text).__name__)
print("pressure_value type:", type(pressure_value).__name__)
print("combined text:", combined_text)
print("converted type:", type(converted_pressure).__name__)
print("safety margin:", f"{safety_margin:.1f} {unit_label}")

```

## **2. User Input Basics**

### **2.1. Input Prompt Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_02_01.jpg?v=1780923397" width="250">



>* Ask users for changing calculation values
>* Use clear prompts before they type

>* Ask clearly for specific values and units
>* Make prompts understandable without reading code

>* Input arrives as text first
>* Clear prompts support accurate conversion



```
#@title Python Code - Input Prompt Basics

# This example demonstrates clear input prompts.
# Engineering prompts should include useful units.
# Mock responses keep Colab execution automatic.

# Store prompt text with explicit units.
prompt_text = "Enter beam length in meters: "

# Simulate what input would return.
typed_response = "2.75"

# Convert the text into numeric data.
beam_length_m = float(typed_response)

# Use the numeric value in calculation.
beam_length_mm = beam_length_m * 1000

# Report both raw and converted information.
print("Prompt shown:", prompt_text)

# Show that typed input starts as text.
print("Typed response:", typed_response)

# Display the original response type.
print("Response type:", type(typed_response).__name__)

# Display the converted numeric type.
print("Converted type:", type(beam_length_m).__name__)

# Format the engineering result clearly.
print(f"Beam length = {beam_length_m:.2f} m = {beam_length_mm:.0f} mm")

```

### **2.2. Converting Numeric Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_02_02.jpg?v=1780923401" width="250">



>* User input starts as text
>* Convert text before engineering calculations

>* Use integers for countable engineering quantities
>* Use floats for measurements with decimals

>* Check entries for invalid numeric formats
>* Guide users before engineering calculations



```
#@title Python Code - Converting Numeric Input

# Numeric input begins as text.
# Conversions enable reliable engineering calculations.
# This example simulates interactive entries.

# Store sample responses from possible users.
raw_length_m = "2.75"
raw_bolt_count = "8"

# Convert text values into numeric types.
length_m = float(raw_length_m)
bolt_count = int(raw_bolt_count)

# Use converted values in calculations.
length_cm = length_m * 100
spacing_cm = length_cm / bolt_count

# Show types after conversion.
print("Original length entry:", raw_length_m, type(raw_length_m))
print("Converted length value:", length_m, type(length_m))
print("Converted bolt count:", bolt_count, type(bolt_count))
print("Length in centimeters:", f"{length_cm:.1f} cm")
print("Bolt spacing estimate:", f"{spacing_cm:.2f} cm")

```

### **2.3. Converting User Input**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_02_03.jpg?v=1780923399" width="250">



>* User input starts as text
>* Convert text before engineering calculations

>* Match conversion to engineering meaning
>* Use counts, measurements, and identifiers correctly

>* Check user entries for messy input
>* Use clear prompts for reliable conversions



```
#@title Python Code - Converting User Input

# Convert text entries into numeric engineering values.
# This example simulates user input safely.
# Numeric types enable reliable engineering calculations.

# Store simulated entries like input returns.
bolt_count_text = " 8 "
beam_length_text = "2.75"
load_text = "1250.5"

# Convert count text into an integer.
bolt_count = int(bolt_count_text.strip())

# Convert measurement text into floating values.
beam_length_m = float(beam_length_text.strip())
applied_load_n = float(load_text.strip())

# Use converted numbers in calculations.
load_per_bolt_n = applied_load_n / bolt_count

# Report values with clear formatting.
print(f"Bolt count type: {type(bolt_count).__name__}")
print(f"Beam length type: {type(beam_length_m).__name__}")
print(f"Total load: {applied_load_n:.1f} N")
print(f"Load per bolt: {load_per_bolt_n:.2f} N")
print(f"Beam length: {beam_length_m:.2f} m")

```

## **3. Formatting Engineering Results**

### **3.1. Float Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_03_01.jpg?v=1780923409" width="250">



>* Control how floating-point results are displayed
>* Show meaningful precision without changing values

>* Match precision to engineering context
>* Make results comparable, trustworthy, and usable

>* Consistent floats make outputs easier to scan
>* Clear formatting supports professional engineering communication



```
#@title Python Code - Float Formatting

# Format floating results for engineering reports.
# Use f-strings to control decimals.
# Compare raw and formatted calculation output.

# Define a simple pipe velocity calculation.
flow_rate_m3_s = 0.0187
pipe_diameter_m = 0.1524
pipe_area_m2 = 3.14159 * (pipe_diameter_m / 2) ** 2

# Calculate average velocity from flow and area.
velocity_m_s = flow_rate_m3_s / pipe_area_m2
pressure_drop_pa = 1287.456789
beam_deflection_in = 0.0372468

# Print raw values before formatting.
print("Raw velocity:", velocity_m_s)
print("Raw pressure drop:", pressure_drop_pa)
print("Raw deflection:", beam_deflection_in)

# Print formatted values with engineering units.
print(f"Velocity: {velocity_m_s:.2f} m/s")
print(f"Pressure drop: {pressure_drop_pa:.1f} Pa")
print(f"Deflection: {beam_deflection_in:.3f} in")

# Align several design alternatives neatly.
print("\nAligned design summary:")
print(f"{'Case':<8}{'Velocity':>12}{'Drop':>12}")
print(f"{'A':<8}{velocity_m_s:>12.2f}{pressure_drop_pa:>12.1f}")
print(f"{'B':<8}{1.18 * velocity_m_s:>12.2f}{1.35 * pressure_drop_pa:>12.1f}")

```

### **3.2. Decimal Precision**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_03_02.jpg?v=1780923413" width="250">



>* Choose decimals to avoid false accuracy
>* Match precision to data and decisions

>* Match precision to engineering context
>* Avoid false accuracy from extra decimals

>* Use consistent decimals for quick comparisons
>* Balance rounding with engineering meaning



```
#@title Python Code - Decimal Precision

# Decimal precision guides engineering communication.
# Formatting should match measurement confidence.
# Python f-strings control displayed decimals.

# Store example engineering calculation results.
length_mm = 25.4
force_newton = 1234.56789
area_square_mm = 78.539816

# Calculate stress using consistent units.
stress_mpa = force_newton / area_square_mm
safety_factor = 2.48765
cost_dollars = 1532.789

# Print raw values for comparison.
print("Raw stress value:", stress_mpa)
print("Formatted engineering results:")
print(f"Part length: {length_mm:.1f} mm")

# Show different practical decimal choices.
print(f"Applied force: {force_newton:.0f} N")
print(f"Area: {area_square_mm:.2f} mm^2")
print(f"Stress: {stress_mpa:.2f} MPa")

# Format safety and cost clearly.
print(f"Safety factor: {safety_factor:.2f}")
print(f"Estimated cost: ${cost_dollars:.2f}")

```

### **3.3. Unsigned Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_01/Lecture_C/image_03_03.jpg?v=1780923411" width="250">



>* Use unsigned values to show magnitude
>* Hide signs only when direction is irrelevant

>* Compare magnitudes against limits and tolerances
>* Show how large deviations are

>* Unsigned values can hide important direction context
>* Use labels and wording for clarity



```
#@title Python Code - Unsigned Formatting

# Demonstrate unsigned formatting for engineering results.
# Show magnitudes without hiding engineering meaning.
# Keep output concise for beginner learners.

# Store signed deviations from reference measurements.
pressure_error_kpa = -2.735
hole_error_mm = 0.084
position_error_in = -0.0126

# Convert signed values into unsigned magnitudes.
pressure_magnitude_kpa = abs(pressure_error_kpa)
hole_magnitude_mm = abs(hole_error_mm)
position_magnitude_in = abs(position_error_in)

# Print signed values for comparison context.
print(f"Signed pressure error: {pressure_error_kpa:+.2f} kPa")
print(f"Unsigned pressure error: {pressure_magnitude_kpa:.2f} kPa")

# Print unsigned values against engineering tolerances.
print(f"Hole size deviation magnitude: {hole_magnitude_mm:.3f} mm")
print(f"Positioning error magnitude: {position_magnitude_in:.4f} in")

# Preserve direction separately when diagnosis matters.
direction_note = "below target" if position_error_in < 0 else "above target"
print(f"Diagnostic note: tool position was {direction_note}")

```
