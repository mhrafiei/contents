## **1. Function Basics**

### **1.1. Defining Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_01_01.jpg?v=1780925487" width="250">



>* Package repeated formulas into reusable functions
>* Use clear names for readable engineering code

>* Give each function one clear job
>* Link formulas to readable, checkable code

>* Plan inputs, outputs, and physical meaning
>* Make formulas clear, consistent, and reusable



```
#@title Python Code - Defining Functions

# Functions package repeated engineering calculations.
# Parameters provide the required formula inputs.
# Return values make results reusable.

# Define a reusable axial stress formula.
def axial_stress(force_newtons, area_square_meters):
    stress_pascals = force_newtons / area_square_meters

    return stress_pascals

# Define a reusable spring deflection formula.
def spring_deflection(load_newtons, stiffness_newtons_per_meter):
    deflection_meters = load_newtons / stiffness_newtons_per_meter

    return deflection_meters

# Define known input values for testing.
force = 12000.0
area = 0.003

# Call the function using named variables.
stress = axial_stress(force, area)
print("Axial stress:", stress, "Pa")

# Test against a simple hand calculation.
expected_stress = 4000000.0
print("Stress test passed:", stress == expected_stress)

# Call another function for a spring case.
load = 500.0
stiffness = 20000.0

deflection = spring_deflection(load, stiffness)
print("Spring deflection:", deflection, "m")

# Compare the result with expected behavior.
expected_deflection = 0.025
print("Deflection test passed:", deflection == expected_deflection)

```

### **1.2. Function Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_01_02.jpg?v=1780925491" width="250">



>* Parameters let functions accept changing engineering values
>* One formula can serve many cases

>* Use parameter names that show quantities clearly
>* Clear names prevent errors and aid teamwork

>* Choose inputs versus fixed assumptions carefully
>* Match function design to physical context



```
#@title Python Code - Function Parameters

# Functions can accept changing engineering inputs.
# Parameters name the needed physical quantities.
# Return values make calculations reusable.

# Define stress using force and area parameters.
def axial_stress(force_newtons, area_square_meters):
    if area_square_meters <= 0:
        raise ValueError("Area must be positive.")

    stress_pascals = force_newtons / area_square_meters
    return stress_pascals

# Define shaft torque using force and radius parameters.
def torque(force_pounds, radius_inches):
    if radius_inches < 0:
        raise ValueError("Radius cannot be negative.")

    torque_pound_inches = force_pounds * radius_inches
    return torque_pound_inches

# Call the same function for different metric cases.
stress_steel = axial_stress(12000, 0.0008)
stress_aluminum = axial_stress(7500, 0.0005)

# Call another function for an imperial example.
shaft_torque = torque(35, 6)

# Check a known example for correct behavior.
expected_stress = 15000000
passed_check = stress_steel == expected_stress

# Print concise results for comparison.
print("Steel stress:", stress_steel, "Pa")
print("Aluminum stress:", stress_aluminum, "Pa")
print("Shaft torque:", shaft_torque, "lb-in")
print("Known stress check passed:", passed_check)

```

### **1.3. Returning Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_01_03.jpg?v=1780925489" width="250">



>* Return values send answers back
>* Separate calculation from output handling

>* Returned values connect engineering calculation steps
>* Clean functions become reliable analysis building blocks

>* Return outputs that match engineering meaning
>* Clear returns support reuse and testing



```
#@title Python Code - Returning Results

# Functions can return reusable engineering results.
# Returned values support later design calculations.
# This example checks simple stress functions.

# Define area returned from circular diameter.
def circular_area(diameter_m):
    area_m2 = 3.141592653589793 * diameter_m**2 / 4
    return area_m2

# Define axial stress returned from load.
def axial_stress(force_N, area_m2):
    stress_Pa = force_N / area_m2
    return stress_Pa

# Define strain returned from deformation.
def engineering_strain(delta_m, length_m):
    strain_value = delta_m / length_m
    return strain_value

# Store known rod inputs.
rod_diameter_m = 0.02
applied_force_N = 5000
original_length_m = 2.0

# Store measured elongation input.
elongation_m = 0.001
rod_area_m2 = circular_area(rod_diameter_m)
rod_stress_Pa = axial_stress(applied_force_N, rod_area_m2)

# Chain returned stress into MPa conversion.
rod_stress_MPa = rod_stress_Pa / 1_000_000
rod_strain = engineering_strain(elongation_m, original_length_m)
expected_area_m2 = 0.0003141592653589793

# Test returned area against known result.
area_test_passed = abs(rod_area_m2 - expected_area_m2) < 1e-12
print(f"Rod area: {rod_area_m2:.6f} m^2")
print(f"Axial stress: {rod_stress_MPa:.2f} MPa")

# Print strain and simple test result.
print(f"Engineering strain: {rod_strain:.6f}")
print(f"Area test passed: {area_test_passed}")

```

## **2. Reusable Formula Functions**

### **2.1. Parameter Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_02_01.jpg?v=1780925502" width="250">



>* Function inputs provide case-specific engineering values
>* Reusable formulas separate logic from examples

>* Choose inputs with clear engineering meaning
>* Clear parameters reduce reuse mistakes

>* Test many designs with one function
>* Build reusable, consistent engineering tools



```
#@title Python Code - Parameter Inputs

# Functions accept values through named parameters.
# Each parameter represents an engineering quantity.
# Returned values make calculations reusable.

# Define stress using force and area.
def normal_stress(force_newtons, area_square_meters):
    stress_pascals = force_newtons / area_square_meters
    return stress_pascals

# Define spring deflection using Hooke's law.
def spring_deflection(force_newtons, stiffness_newtons_per_meter):
    deflection_meters = force_newtons / stiffness_newtons_per_meter
    return deflection_meters

# Evaluate the same stress function twice.
steel_stress = normal_stress(12000.0, 0.0030)
aluminum_stress = normal_stress(8500.0, 0.0025)
steel_stress_mpa = steel_stress / 1_000_000.0
aluminum_stress_mpa = aluminum_stress / 1_000_000.0

# Evaluate the same spring function twice.
soft_spring = spring_deflection(400.0, 20000.0)
stiff_spring = spring_deflection(400.0, 50000.0)
soft_spring_mm = soft_spring * 1000.0
stiff_spring_mm = stiff_spring * 1000.0

# Test one known calculation silently.
known_stress = normal_stress(1000.0, 0.0010)
assert abs(known_stress - 1_000_000.0) < 1e-9

# Print compact engineering results.
print("Reusable parameter input examples:")
print(f"Steel bracket stress: {steel_stress_mpa:.2f} MPa")
print(f"Aluminum panel stress: {aluminum_stress_mpa:.2f} MPa")
print(f"Soft spring deflection: {soft_spring_mm:.1f} mm")
print(f"Stiff spring deflection: {stiff_spring_mm:.1f} mm")
print("Known stress test passed.")

```

### **2.2. Parameter Return Flow**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_02_02.jpg?v=1780925498" width="250">



>* Functions take inputs and calculate results
>* Returned results can be reused later

>* Parameters provide inputs; returns provide results
>* Reusable functions adapt to different engineering cases

>* Returned results connect later engineering checks
>* Focused functions combine into complete analyses



```
#@title Python Code - Parameter Return Flow

# Functions make engineering formulas reusable.
# Parameters carry case information inward.
# Return values carry results outward.

# Define a stress function using parameters.
def axial_stress_mpa(force_newtons, area_mm2):
    stress_mpa = force_newtons / area_mm2
    return stress_mpa

# Define a torque function using parameters.
def shaft_torque_nm(force_newtons, radius_meters):
    torque_nm = force_newtons * radius_meters
    return torque_nm

# Define a safety factor function.
def safety_factor(allowable_mpa, actual_mpa):
    factor_value = allowable_mpa / actual_mpa
    return factor_value

# Store inputs for one metric case.
load_n = 12000
area_mm2 = 300
allowable_mpa = 250

# Call functions and store returns.
stress_result = axial_stress_mpa(load_n, area_mm2)
torque_result = shaft_torque_nm(load_n, 0.08)
factor_result = safety_factor(allowable_mpa, stress_result)

# Show the parameter return flow.
print("Inputs entered: force and area")
print(f"Returned stress: {stress_result:.1f} MPa")
print(f"Returned torque: {torque_result:.1f} N m")
print(f"Returned safety factor: {factor_result:.2f}")

# Reuse the same function differently.
second_stress = axial_stress_mpa(8000, 200)
print(f"Second returned stress: {second_stress:.1f} MPa")

```

### **2.3. Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_02_03.jpg?v=1780925500" width="250">



>* Returned results can be reused later
>* Outputs connect formulas to engineering workflows

>* Consistent outputs make formulas reusable
>* Returned results support reliable design comparisons

>* Return one clear engineering result
>* Use outputs in later calculation chains



```
#@title Python Code - Return Values

# Functions can return reusable engineering results.
# Returned values support later design checks.
# This example uses stress and safety.

# Define stress using force and area.
def calculate_stress(force_newtons, area_square_meters):
    stress_pascals = force_newtons / area_square_meters
    return stress_pascals

# Define safety factor from allowable stress.
def calculate_safety_factor(allowable_pascals, actual_pascals):
    safety_factor = allowable_pascals / actual_pascals
    return safety_factor

# Store inputs for a simple bracket.
force_newtons = 2500.0
area_square_meters = 0.0005
allowable_pascals = 120_000_000.0

# Call functions and keep returned values.
stress_pascals = calculate_stress(force_newtons, area_square_meters)
safety_factor = calculate_safety_factor(allowable_pascals, stress_pascals)
stress_megapascals = stress_pascals / 1_000_000.0

# Compare the returned value directly.
design_is_safe = safety_factor >= 2.0
print(f"Returned stress: {stress_megapascals:.2f} MPa")
print(f"Returned safety factor: {safety_factor:.2f}")
print(f"Meets minimum safety factor of 2: {design_is_safe}")

```

## **3. Function Testing**

### **3.1. Known Value Checks**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_03_01.jpg?v=1780925494" width="250">



>* Test functions using already-known answers
>* Compare results with engineering expectations

>* Choose simple, meaningful test values deliberately
>* Check arithmetic and physical behavior

>* Test normal cases and edge conditions
>* Build trust before engineering decisions



```
#@title Python Code - Known Value Checks

# Known values help verify engineering functions.
# Simple cases reveal formula mistakes quickly.
# Edge cases confirm sensible physical behavior.

# Define a reusable cylindrical volume function.
def cylinder_volume(radius_m, height_m):
    return 3.141592653589793 * radius_m**2 * height_m

# Define a reusable thermal expansion function.
def thermal_expansion(length_m, alpha_per_c, delta_temp_c):
    return length_m * alpha_per_c * delta_temp_c

# Store known checks as labeled tuples.
volume_check = ("Volume", cylinder_volume(1.0, 2.0), 6.283185307179586)
expansion_check = ("Zero expansion", thermal_expansion(2.0, 12e-6, 0.0), 0.0)

# Compare computed values with trusted known values.
def check_known_value(label, calculated, expected, tolerance):
    error = abs(calculated - expected)
    status = "PASS" if error <= tolerance else "FAIL"
    return f"{label}: {status}, error = {error:.3e}"

# Run ordinary and edge case checks.
print(check_known_value(volume_check[0], volume_check[1], volume_check[2], 1e-12))
print(check_known_value(expansion_check[0], expansion_check[1], expansion_check[2], 1e-12))

# Show one familiar imperial stress calculation.
def stress_psi(force_lb, area_in2):
    return force_lb / area_in2

# Check stress using easy hand calculation.
stress_message = check_known_value("Stress", stress_psi(1000.0, 2.0), 500.0, 1e-12)
print(stress_message)

```

### **3.2. Expected Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_03_02.jpg?v=1780925496" width="250">



>* Know correct answers before testing functions
>* Compare outputs to judge engineering accuracy

>* Choose simple, meaningful checks for calculations
>* Catch unit, input, and logic mistakes

>* Known checks build reusable function trust
>* Test normal, edge, and unrealistic cases



```
#@title Python Code - Expected Results

# Expected results guide engineering function testing.
# Known answers help reveal formula mistakes.
# Small tests build reusable calculation confidence.

# Define a reusable force calculation function.
def force_newtons(mass_kg, acceleration_m_s2):
    return mass_kg * acceleration_m_s2

# Define a simple beam stress function.
def stress_mpa(force_newtons_value, area_mm2):
    stress_value = force_newtons_value / area_mm2
    return stress_value

# Define clear expected results before testing.
expected_force = 50.0
calculated_force = force_newtons(10.0, 5.0)
force_passed = abs(calculated_force - expected_force) < 1e-9

# Test stress using compatible metric units.
expected_stress = 25.0
calculated_stress = stress_mpa(5000.0, 200.0)
stress_passed = abs(calculated_stress - expected_stress) < 1e-9

# Show compact engineering test results.
print("Function testing with expected results")
print(f"Force test: expected {expected_force} N, got {calculated_force} N")
print(f"Force test passed: {force_passed}")
print(f"Stress test: expected {expected_stress} MPa, got {calculated_stress} MPa")
print(f"Stress test passed: {stress_passed}")

```

### **3.3. Expected Results**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_B/image_03_03.jpg?v=1780925492" width="250">



>* Set expected answers before trusting functions
>* Compare results against validated engineering examples

>* Choose realistic cases to reveal calculation mistakes
>* Estimate results before comparing physical plausibility

>* Reuse expected results to catch regressions
>* Build reliable habits for engineering calculations



```
#@title Python Code - Expected Results

# Functions should match known engineering examples.
# Expected results confirm formula translation.
# Tolerance allows tiny floating point differences.

# Define axial stress using force and area.
def axial_stress(force_newtons, area_square_meters):
    return force_newtons / area_square_meters

# Define percent difference for test reporting.
def percent_error(calculated_value, expected_value):
    difference = abs(calculated_value - expected_value)
    return 100 * difference / abs(expected_value)

# Store one known metric stress example.
force_newtons = 10000
area_square_meters = 0.005
expected_stress = 2000000

# Calculate the function result for comparison.
calculated_stress = axial_stress(force_newtons, area_square_meters)
stress_error = percent_error(calculated_stress, expected_stress)
tolerance_percent = 0.01

# Check whether the result meets expectation.
stress_passed = stress_error <= tolerance_percent
print("Axial stress test passed:", stress_passed)
print("Calculated stress in pascals:", calculated_stress)

# Define beam deflection for central point load.
def center_deflection(load, length, modulus, inertia):
    denominator_value = 48 * modulus * inertia
    return load * length**3 / denominator_value

# Store a simple handbook-style beam example.
load_newtons = 500
length_meters = 2
modulus_pascals = 200_000_000_000
inertia_meters4 = 8.333e-6

# Compare against a precomputed expected deflection.
expected_deflection = 0.000125005
calculated_deflection = center_deflection(load_newtons, length_meters, modulus_pascals, inertia_meters4)
deflection_error = percent_error(calculated_deflection, expected_deflection)

# Print compact testing evidence for learners.
deflection_passed = deflection_error < 0.01
print("Beam deflection test passed:", deflection_passed)
print("Calculated deflection in meters:", round(calculated_deflection, 9))

```
