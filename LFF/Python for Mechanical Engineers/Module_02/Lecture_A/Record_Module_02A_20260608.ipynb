## **1. Conditional Stress Checks**

### **1.1. Allowable Stress Comparison**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_01_01.jpg?v=1780924285" width="250">



>* Compare calculated stress with allowable limits
>* Use conditions to mark pass or fail

>* Compare calculated stress with allowable capacity
>* Pass-fail checks reduce unsafe ambiguity

>* Set clear pass-fail stress boundaries
>* Check other risks before accepting designs



```
#@title Python Code - Allowable Stress Comparison

# Compare stress with allowable limits.
# Conditional statements create engineering decisions.
# This example checks one bracket.

# Define the applied tensile load.
load_newton = 12000.0

# Define the resisting cross sectional area.
area_mm2 = 85.0

# Define the allowable normal stress.
allowable_mpa = 150.0

# Calculate stress using compatible units.
stress_mpa = load_newton / area_mm2

# Calculate remaining stress margin.
margin_mpa = allowable_mpa - stress_mpa

# Print the computed stress value.
print(f"Calculated stress: {stress_mpa:.1f} MPa")

# Print the allowable stress value.
print(f"Allowable stress: {allowable_mpa:.1f} MPa")

# Use a conditional statement.
if stress_mpa <= allowable_mpa:
    result_text = "PASS: stress is within the allowable limit"
else:
    result_text = "FAIL: stress exceeds the allowable limit"

# Print the engineering decision.
print(result_text)

# Print the signed stress margin.
print(f"Stress margin: {margin_mpa:.1f} MPa")

```

### **1.2. Allowable Stress Limits**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_01_02.jpg?v=1780924283" width="250">



>* Limits separate acceptable designs from unsafe ones
>* Safety margins account for real-world uncertainty

>* Compare stress to decide pass or fail
>* Adjust limits for fatigue and temperature

>* Limits reflect judgment, not guaranteed safety
>* Choose thresholds balancing risk, cost, performance



```
#@title Python Code - Allowable Stress Limits

# Compare stress against allowable limits.
# Use clear engineering decision logic.
# Keep calculations small and readable.

# Define computed bracket stress values.
stress_mpa = [115, 148, 151, 190]

# Define one allowable stress limit.
allowable_mpa = 150

# Check each stress against limit.
for case_number, stress_value in enumerate(stress_mpa, start=1):
    # Decide whether this case passes.
    if stress_value <= allowable_mpa:
        status_text = "PASS"

    else:
        status_text = "REDESIGN"

    # Print one concise design decision.
    print(f"Case {case_number}: {stress_value} MPa -> {status_text}")

# Summarize the allowable design threshold.
print(f"Allowable stress limit: {allowable_mpa} MPa")

```

### **1.3. Allowable Stress Logic**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_01_03.jpg?v=1780924287" width="250">



>* Compare calculated stress with allowable stress
>* Classify designs as acceptable or overstressed

>* Compare stress using safety meaning
>* Treat equal stress as passing

>* Turn stress results into clear status
>* Flag near-limit cases for engineering action



```
#@title Python Code - Allowable Stress Logic

# Conditional checks compare stress with limits.
# Boundary equality usually passes engineering checks.
# Clear status labels support design decisions.

# Define computed stresses in megapascals.
stresses_mpa = [85.0, 120.0, 151.0]

# Define one allowable stress limit.
allowable_mpa = 150.0

# Print a compact heading for learners.
print("Allowable stress check for three cases")

# Check each stress against the limit.
for case_number, stress_mpa in enumerate(stresses_mpa, start=1):

    # Treat equality as acceptable here.
    if stress_mpa <= allowable_mpa:

        # Label cases within the allowable limit.
        status = "ACCEPTABLE"

    # Handle stresses above the allowable limit.
    else:

        # Label cases exceeding the allowable limit.
        status = "OVERSTRESSED"

    # Report each decision in one line.
    print(f"Case {case_number}: {stress_mpa:.1f} MPa -> {status}")

# Show the engineering rule being applied.
print(f"Rule: pass when stress <= {allowable_mpa:.1f} MPa")

```

## **2. Loop Patterns**

### **2.1. For Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_02_01.jpg?v=1780924280" width="250">



>* Repeat known engineering calculations automatically
>* Process each load, speed, or temperature consistently

>* Repeat fixed analyses consistently and reliably
>* Compare design cases and reveal trends

>* Separate cases from repeated actions.
>* Make analyses clear, consistent, and reviewable.



```
#@title Python Code - For Loops

# For loops repeat calculations consistently.
# Mechanical load cases often form lists.
# Each loop pass handles one case.

# Define bracket load cases in newtons.
loads_newton = [500, 750, 1000, 1250]

# Define a constant bracket area.
area_mm2 = 250

# Define the allowable stress limit.
allowable_mpa = 4.0

# Store calculated stresses for review.
stresses_mpa = []

# Loop through each load case.
for load_newton in loads_newton:

    # Calculate stress using megapascals.
    stress_mpa = load_newton / area_mm2

    # Save each calculated stress value.
    stresses_mpa.append(stress_mpa)

# Print a compact table heading.
print("Load N | Stress MPa | Decision")

# Loop through paired loads and stresses.
for load_newton, stress_mpa in zip(loads_newton, stresses_mpa):

    # Classify each engineering case.
    if stress_mpa <= allowable_mpa:

        # Mark acceptable design cases.
        decision = "OK"

    # Handle cases above the limit.
    else:

        # Mark cases exceeding allowable stress.
        decision = "Too high"

    # Print one concise result row.
    print(f"{load_newton:6.0f} | {stress_mpa:10.2f} | {decision}")

```

### **2.2. While Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_02_02.jpg?v=1780924277" width="250">



>* Repeat calculations until a condition changes
>* Useful when attempts are not predetermined

>* Start, test, and update toward completion
>* Use safeguards to avoid endless loops

>* Use while loops for design improvement searches
>* Stop at realistic, meaningful acceptable cases



```
#@title Python Code - While Loops

# While loops repeat until conditions change.
# This example sizes a simple shaft.
# Safety limits prevent endless design searching.

# Define torque and allowable shear stress.
torque_in_lb = 1200.0
allowable_psi = 8000.0

# Start with a small trial diameter.
diameter_in = 0.25
step_in = 0.05

# Set practical and counting safeguards.
max_diameter_in = 2.00
attempt_count = 0

# Compute initial shear stress estimate.
stress_psi = 16.0 * torque_in_lb
stress_psi = stress_psi / (3.14159 * diameter_in**3)

# Increase diameter while stress remains excessive.
while stress_psi > allowable_psi and diameter_in <= max_diameter_in:
    attempt_count = attempt_count + 1
    diameter_in = diameter_in + step_in

    # Recalculate stress after changing diameter.
    stress_psi = 16.0 * torque_in_lb
    stress_psi = stress_psi / (3.14159 * diameter_in**3)

# Decide whether the search found success.
if stress_psi <= allowable_psi:
    result_text = "acceptable shaft diameter found"
else:
    result_text = "no acceptable diameter within limit"

# Print a compact engineering summary.
print("While-loop shaft sizing example")
print("Result:", result_text)
print("Attempts:", attempt_count)
print("Diameter, in:", round(diameter_in, 3))
print("Shear stress, psi:", round(stress_psi, 1))

```

### **2.3. Array Iteration**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_02_03.jpg?v=1780924281" width="250">



>* Loop through engineering data consistently
>* Apply calculations to each array value

>* Compare many engineering cases consistently
>* Scale checks to large datasets safely

>* Know each element’s purpose
>* Keep related arrays correctly matched



```
#@title Python Code - Array Iteration

# Iterate through engineering measurement arrays.
# Compare each result with limits.
# Keep outputs short for learning.

# Store matching torque measurements and speeds.
torque_nm = [80, 120, 160, 200]
speed_rpm = [900, 1200, 1500, 1800]

# Set allowable shaft power threshold.
allowable_kw = 25.0
results = []

# Check array alignment before iterating.
if len(torque_nm) != len(speed_rpm):
    raise ValueError("Torque and speed arrays must align")

# Visit paired array values together.
for torque, speed in zip(torque_nm, speed_rpm):
    power_kw = torque * speed * 2 * 3.14159 / 60000
    status = "OK" if power_kw <= allowable_kw else "HIGH"

    # Record each computed design case.
    results.append((torque, speed, power_kw, status))

# Print a compact engineering table.
print("Torque N*m | Speed rpm | Power kW | Status")

# Display each processed array element.
for torque, speed, power_kw, status in results:
    print(f"{torque:10.0f} | {speed:9.0f} | {power_kw:8.2f} | {status}")

# Summarize the repeated calculation.
high_count = sum(1 for item in results if item[3] == "HIGH")
print(f"Cases above limit: {high_count} of {len(results)}")

```

## **3. Design Case Classification**

### **3.1. Failure Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_03_01.jpg?v=1780924289" width="250">



>* Classify each design case using loops
>* Turn calculations into clear engineering decisions

>* Classify risk beyond simple pass or fail
>* Use loop decisions to flag redesign cases

>* Clear rules identify key failure causes
>* Classifications reveal patterns across design cases



```
#@title Python Code - Failure Classification

# Classify shaft design cases using loops.
# Compare computed stresses against allowable limits.
# Print compact engineering judgment for each case.

# Define material limits in megapascals.
allowable_stress_mpa = 140.0
marginal_ratio = 0.90

# Store torque cases in newton meters.
torque_cases_nm = [80, 110, 130, 150, 180]
shaft_diameter_m = 0.025

# Prepare a short classification summary.
summary_counts = {"acceptable": 0, "marginal": 0, "failed": 0}
print("Torque case failure classification")

# Loop through each operating torque case.
for case_number, torque_nm in enumerate(torque_cases_nm, start=1):
    polar_moment_m4 = 3.14159 * shaft_diameter_m**4 / 32
    radius_m = shaft_diameter_m / 2

    # Calculate maximum torsional shear stress.
    stress_pa = torque_nm * radius_m / polar_moment_m4
    stress_mpa = stress_pa / 1_000_000

    # Classify by comparing stress with limits.
    if stress_mpa > allowable_stress_mpa:
        status = "failed"
    elif stress_mpa >= marginal_ratio * allowable_stress_mpa:
        status = "marginal"
    else:
        status = "acceptable"

    # Store one count for each status.
    summary_counts[status] = summary_counts[status] + 1
    print(f"Case {case_number}: {torque_nm} N·m, {stress_mpa:.1f} MPa, {status}")

# Print the compact design summary.
print("Summary:", summary_counts)

```

### **3.2. Case Classification**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_03_02.jpg?v=1780924293" width="250">



>* Repeat checks across many design cases
>* Label results for clearer engineering decisions

>* Define useful categories and consistent rules
>* Classify cases for fair engineering decisions

>* Classify many cases quickly and consistently
>* Reveal patterns for repeatable design screening



```
#@title Python Code - Case Classification

# Classify multiple shaft design cases.
# Loops repeat calculations consistently.
# Decisions turn results into labels.

# Define allowable engineering limits.
allowable_stress_mpa = 150.0
max_deflection_mm = 0.60
preferred_mass_kg = 4.50

# Store several small design cases.
cases = [
    {"name": "A", "stress": 118.0, "deflection": 0.42, "mass": 3.9},
    {"name": "B", "stress": 146.0, "deflection": 0.58, "mass": 4.7},
    {"name": "C", "stress": 171.0, "deflection": 0.51, "mass": 4.2},
]

# Add another borderline design case.
cases.append({"name": "D", "stress": 139.0, "deflection": 0.72, "mass": 5.1})

# Validate the small case list.
if len(cases) == 0:
    raise ValueError("At least one case is required")

# Print a compact table heading.
print("Case | stress MPa | defl mm | class")

# Classify each case using identical logic.
for case in cases:
    stress_ratio = case["stress"] / allowable_stress_mpa
    deflection_ratio = case["deflection"] / max_deflection_mm

    # Start with a safe default label.
    label = "acceptable"
    if stress_ratio > 1.0 or deflection_ratio > 1.0:
        label = "redesign required"

    # Mark costly but safe designs.
    elif case["mass"] > preferred_mass_kg:
        label = "safe but heavy"

    # Mark cases approaching limits.
    elif stress_ratio > 0.90 or deflection_ratio > 0.90:
        label = "marginal"

    # Print one concise result line.
    print(case["name"], "|", case["stress"], "|", case["deflection"], "|", label)

# Count cases needing design attention.
attention_count = 0
for case in cases:
    if case["stress"] > allowable_stress_mpa:
        attention_count = attention_count + 1

# Include deflection failures in attention count.
for case in cases:
    if case["deflection"] > max_deflection_mm:
        attention_count = attention_count + 1

# Print the screening summary.
print("Limit violations counted:", attention_count)

```

### **3.3. Summarizing Classifications**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_02/Lecture_A/image_03_03.jpg?v=1780924291" width="250">



>* Summaries reveal patterns across design cases
>* They show pass limits and failure reasons

>* Summarize key counts and failure patterns
>* Reveal trends, trade-offs, and critical inputs

>* Connect computed cases to engineering judgment
>* Reveal patterns guiding safer design choices



```
#@title Python Code - Summarizing Classifications

# Summaries turn classifications into engineering design insight.
# Loops count categories across many design cases.
# Decisions explain why each case was classified.

# Define allowable limits for a bracket study.
allowable_stress_mpa = 160
allowable_deflection_mm = 4.0
allowable_temperature_c = 90

# Store small cases with mechanical design inputs.
cases = [
    {"name": "Case A", "stress": 120, "deflection": 2.8, "temperature": 65},
    {"name": "Case B", "stress": 155, "deflection": 3.7, "temperature": 82},
    {"name": "Case C", "stress": 178, "deflection": 3.1, "temperature": 70},
    {"name": "Case D", "stress": 145, "deflection": 5.2, "temperature": 76},
    {"name": "Case E", "stress": 138, "deflection": 3.5, "temperature": 98},
    {"name": "Case F", "stress": 170, "deflection": 4.8, "temperature": 96},
]

# Prepare summary counters before classification begins.
summary = {"acceptable": 0, "marginal": 0, "failed": 0}
reasons = {"stress": 0, "deflection": 0, "temperature": 0}

# Classify each case using engineering decisions.
for case in cases:
    stress_ratio = case["stress"] / allowable_stress_mpa
    deflection_ratio = case["deflection"] / allowable_deflection_mm
    temperature_ratio = case["temperature"] / allowable_temperature_c

    # Find the largest demand ratio for this case.
    ratios = {
        "stress": stress_ratio,
        "deflection": deflection_ratio,
        "temperature": temperature_ratio,
    }

    controlling_reason = max(ratios, key=ratios.get)
    controlling_ratio = ratios[controlling_reason]

    # Assign category from the controlling demand ratio.
    if controlling_ratio <= 0.85:
        category = "acceptable"
    elif controlling_ratio <= 1.00:
        category = "marginal"
    else:
        category = "failed"

    summary[category] += 1
    reasons[controlling_reason] += 1

# Compute percentages for concise engineering reporting.
total_cases = len(cases)
acceptable_percent = 100 * summary["acceptable"] / total_cases
marginal_percent = 100 * summary["marginal"] / total_cases
failed_percent = 100 * summary["failed"] / total_cases

# Identify the most common controlling design limit.
dominant_reason = max(reasons, key=reasons.get)

# Print a compact classification summary.
print("Design case classification summary")
print(f"Total cases checked: {total_cases}")
print(f"Acceptable: {summary['acceptable']} cases, {acceptable_percent:.1f}%")
print(f"Marginal: {summary['marginal']} cases, {marginal_percent:.1f}%")
print(f"Failed: {summary['failed']} cases, {failed_percent:.1f}%")
print(f"Dominant controlling limit: {dominant_reason}")

# Provide one engineering interpretation from the summary.
if summary["failed"] > summary["acceptable"]:
    message = "Review the design before moving toward production."
elif summary["marginal"] > 0:
    message = "Investigate marginal cases near allowable limits."
else:
    message = "All cases show comfortable design margin."

print(message)

```
