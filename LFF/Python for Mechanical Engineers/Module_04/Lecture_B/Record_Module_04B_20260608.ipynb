### **1.1. Writing Results**

>* Keep records clear, concise, and complete



```
#@title Python Code - Writing Results

# Write engineering results to readable report files.
# Clear labels and units prevent confusion.
# This example saves a beam deflection summary.

# Define simple beam calculation input values.
load_newtons = 1200.0
span_meters = 2.4
elastic_modulus_pa = 200.0e9

# Define rectangular beam section dimensions.
width_meters = 0.05
height_meters = 0.10
report_file = "beam_deflection_report.txt"

# Compute the rectangular second moment of area.
inertia_m4 = width_meters * height_meters**3 / 12.0

# Compute simply supported center deflection.
deflection_meters = load_newtons * span_meters**3

deflection_meters /= 48.0 * elastic_modulus_pa * inertia_m4

# Convert deflection into millimeters for readability.
deflection_mm = deflection_meters * 1000.0
stress_mpa = load_newtons * span_meters * height_meters

# Estimate maximum bending stress for the beam.
stress_mpa /= 4.0 * inertia_m4 * 1.0e6

# Write labeled results and units to a text file.
with open(report_file, "w", encoding="utf-8") as file_handle:
    file_handle.write("Beam Deflection Report\n")
    file_handle.write("Units are shown with each result.\n")
    file_handle.write(f"Applied load: {load_newtons:.1f} N\n")

    file_handle.write(f"Beam span: {span_meters:.2f} m\n")
    file_handle.write(f"Width: {width_meters:.3f} m\n")
    file_handle.write(f"Height: {height_meters:.3f} m\n")
    file_handle.write(f"Moment of inertia: {inertia_m4:.3e} m^4\n")

    file_handle.write(f"Maximum deflection: {deflection_mm:.3f} mm\n")
    file_handle.write(f"Maximum bending stress: {stress_mpa:.2f} MPa\n")

# Read the saved report to verify its contents.
with open(report_file, "r", encoding="utf-8") as file_handle:
    report_text = file_handle.read()

# Display the report filename and readable contents.
print(f"Saved report file: {report_file}")
print(report_text)

```

### **2.1. Importing Custom Modules**

```
#@title Python Code - Importing Custom Modules

# Custom modules keep engineering calculations organized.
# This example imports functions from another file.
# The workflow writes a small labeled report.

# Import pathlib for safe local file paths.
from pathlib import Path

# Store the custom module filename.
module_path = Path("beam_tools.py")

# Write reusable functions into a module file.
module_code = '''
def rectangular_area(width_in, height_in):
    return width_in * height_in

def bending_stress(moment_lb_in, width_in, height_in):
    section_modulus = width_in * height_in**2 / 6
    return moment_lb_in / section_modulus
'''

# Create the module beside this notebook script.
module_path.write_text(module_code)

# Import the custom module by filename stem.
import beam_tools

# Define one tiny engineering input dataset.
width_in = 2.0
height_in = 4.0
moment_lb_in = 1200.0

# Compute results using imported module functions.
area_in2 = beam_tools.rectangular_area(width_in, height_in)
stress_psi = beam_tools.bending_stress(moment_lb_in, width_in, height_in)

# Prepare a clear report with labels and units.
report_lines = [
    "Beam calculation report",
    f"Width: {width_in:.2f} in",
    f"Height: {height_in:.2f} in",
    f"Area: {area_in2:.2f} in^2",
    f"Bending stress: {stress_psi:.1f} psi",
]

# Write the report to a small text file.
report_path = Path("beam_report.txt")
report_path.write_text("\n".join(report_lines))

# Print a compact summary for learners.
print("Imported module: beam_tools.py")
print(f"Area computed from module: {area_in2:.2f} in^2")
print(f"Stress computed from module: {stress_psi:.1f} psi")
print(f"Report written to: {report_path}")

```

### **2.2. Reusable Function Modules**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_B/image_02_02.jpg?v=1780930791" width="250">



>* Store related calculations in reusable module files
>* Keep main programs clearer and workflow-focused

>* Group related functions with clear names
>* Reuse tested code to reduce errors

>* Separate calculations for easier testing
>* Build maintainable libraries with clear responsibilities



```
#@title Python Code - Reusable Function Modules

# Reusable modules keep engineering formulas organized.
# This example creates and imports one module.
# Results stay labeled with clear engineering units.

# Import standard tools for files.
from pathlib import Path
import importlib.util

# Choose a small local module file.
module_path = Path("beam_calcs.py")

# Store reusable beam functions as text.
module_code = '''
def bending_stress(moment_n_m, distance_m, inertia_m4):
    if inertia_m4 <= 0:
        raise ValueError("inertia_m4 must be positive")
    return moment_n_m * distance_m / inertia_m4

def tip_deflection(load_n, length_m, modulus_pa, inertia_m4):
    if modulus_pa <= 0 or inertia_m4 <= 0:
        raise ValueError("modulus_pa and inertia_m4 must be positive")
    return load_n * length_m**3 / (3 * modulus_pa * inertia_m4)
'''

# Write the reusable module file.
module_path.write_text(module_code)

# Prepare an import specification.
spec = importlib.util.spec_from_file_location("beam_calcs", module_path)

# Create a module object safely.
beam_calcs = importlib.util.module_from_spec(spec)

# Load functions from the module file.
spec.loader.exec_module(beam_calcs)

# Define one small cantilever beam case.
load_n = 500.0
length_m = 1.2
modulus_pa = 200e9
inertia_m4 = 8.0e-8

# Define distance from neutral axis.
distance_m = 0.025
moment_n_m = load_n * length_m

# Call reusable functions from module.
stress_pa = beam_calcs.bending_stress(moment_n_m, distance_m, inertia_m4)
deflection_m = beam_calcs.tip_deflection(load_n, length_m, modulus_pa, inertia_m4)

# Convert results into friendly units.
stress_mpa = stress_pa / 1_000_000
deflection_mm = deflection_m * 1000

# Write a tiny labeled report.
report_path = Path("beam_report.txt")
report_lines = [
    "Cantilever beam report",
    f"Load: {load_n:.1f} N",
    f"Length: {length_m:.2f} m",
    f"Bending stress: {stress_mpa:.2f} MPa",
    f"Tip deflection: {deflection_mm:.3f} mm",
]

# Save the report as plain text.
report_path.write_text("\n".join(report_lines))

# Print only the key workflow outputs.
print(f"Module created: {module_path.name}")
print(f"Report created: {report_path.name}")
print(f"Stress = {stress_mpa:.2f} MPa")
print(f"Deflection = {deflection_mm:.3f} mm")

```

```
#@title Python Code - Importing Custom Modules

# This example imports a custom engineering module.
# The module stores reusable beam calculation functions.
# The main script stays focused on workflow.

# Import standard tools for file creation.
from pathlib import Path
import importlib

# Create a small project folder.
project_folder = Path("beam_project")
project_folder.mkdir(exist_ok=True)

# Define the custom module contents.
module_text = """
def rectangular_area(width_m, height_m):
    return width_m * height_m

def cantilever_tip_deflection(load_n, length_m, modulus_pa, inertia_m4):
    return load_n * length_m**3 / (3 * modulus_pa * inertia_m4)
"""

# Save the module beside the script.
module_path = project_folder / "beam_tools.py"
module_path.write_text(module_text)

# Make the folder importable during execution.
import sys
sys.path.insert(0, str(project_folder.resolve()))

# Import the custom calculation module.
beam_tools = importlib.import_module("beam_tools")

# Store input values with metric units.
width_m = 0.050
height_m = 0.100
load_n = 250.0

# Store remaining beam properties.
length_m = 1.20
modulus_pa = 200e9
inertia_m4 = width_m * height_m**3 / 12

# Call reusable functions from the module.
area_m2 = beam_tools.rectangular_area(width_m, height_m)
deflection_m = beam_tools.cantilever_tip_deflection(load_n, length_m, modulus_pa, inertia_m4)

# Write a concise labeled engineering report.
report_path = project_folder / "beam_report.txt"
report_path.write_text(f"Area = {area_m2:.6f} m^2\nTip deflection = {deflection_m*1000:.3f} mm\n")

# Print the workflow results clearly.
print("Custom module imported: beam_tools.py")
print(f"Area = {area_m2:.6f} m^2")
print(f"Tip deflection = {deflection_m*1000:.3f} mm")
print(f"Report written to: {report_path}")

```

## **3. Automated Engineering Workflow**

### **3.1. Workflow Report**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_B/image_03_01.jpg?v=1780930793" width="250">



>* Show how inputs become results
>* Label values, units, and context

>* Use consistent, readable report sections
>* Label units clearly to prevent errors

>* Trace inputs and calculations for review
>* Save reusable records for accountable decisions



```
#@title Python Code - Workflow Report

# Build a repeatable engineering workflow.
# Read inputs, compute values, report.
# Keep labels and units visible.

from pathlib import Path

# Store tiny input data as text.
input_path = Path("pump_inputs.csv")
input_text = "case,flow_gpm,head_ft,power_hp\nA,120,85,5.1\nB,150,78,5.8\n"
input_path.write_text(input_text)

# Define a reusable calculation function.
def pump_efficiency(flow_gpm, head_ft, power_hp):
    water_hp = flow_gpm * head_ft / 3960
    efficiency = water_hp / power_hp
    return efficiency * 100

# Read the prepared workflow inputs.
rows = input_path.read_text().strip().splitlines()
headers = rows[0].split(",")
data_rows = [row.split(",") for row in rows[1:]]

# Validate that each row matches headers.
valid_rows = [row for row in data_rows if len(row) == len(headers)]
if len(valid_rows) != len(data_rows):
    raise ValueError("Input rows do not match headers")

# Compute labeled engineering results.
report_lines = ["Pump Workflow Report", "Input file: pump_inputs.csv", "Fluid: water"]
for row in valid_rows:
    case, flow, head, power = row
    eta = pump_efficiency(float(flow), float(head), float(power))
    report_lines.append(f"Case {case}: efficiency = {eta:.1f} percent")

# Write the final workflow report.
report_path = Path("pump_workflow_report.txt")
report_text = "\n".join(report_lines)
report_path.write_text(report_text)

# Show concise visible workflow output.
print("Workflow complete: report written.")
print(f"Input cases processed: {len(valid_rows)}")
print(f"Report file: {report_path}")
print(report_text)

```

### **3.2. Workflow Report**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_B/image_03_02.jpg?v=1780930797" width="250">



>* Reports summarize inputs, calculations, and findings
>* Clear labels and units build trust

>* Consistent reports reduce mistakes and omissions
>* Stable formats support review and documentation

>* Write reports for future readers
>* Make data, calculations, and units clear



```
#@title Python Code - Workflow Report

# This example creates a repeatable workflow.
# It reads input, computes results, writes report.
# Clear labels and units support engineering review.

from pathlib import Path
import csv

# Define small workflow file locations.
input_path = Path("pump_test_input.csv")
report_path = Path("pump_workflow_report.txt")

# Store tiny sample measurements with units.
rows = [
    ["test_id", "flow_gpm", "head_ft", "efficiency_percent"],
    ["A", "120", "58", "71"],
    ["B", "135", "61", "74"],
    ["C", "128", "60", "73"],
]

# Write input data for repeatable runs.
with input_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Read measurements from the input file.
with input_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Validate that the workflow has data.
if len(data) == 0:
    raise ValueError("Input file contains no measurements.")

# Convert text values into numeric engineering data.
flows = [float(row["flow_gpm"]) for row in data]
heads = [float(row["head_ft"]) for row in data]
effs = [float(row["efficiency_percent"]) for row in data]

# Compute summary values for the report.
avg_flow = sum(flows) / len(flows)
max_head = max(heads)
avg_eff = sum(effs) / len(effs)

# Build readable report lines with units.
report_lines = [
    "Pump Test Workflow Report",
    "Input source: pump_test_input.csv",
    f"Number of tests: {len(data)}",
    f"Average flow rate: {avg_flow:.1f} gpm",
    f"Maximum pump head: {max_head:.1f} ft",
    f"Average efficiency: {avg_eff:.1f} percent",
    "Assumption: all measurements use steady operation.",
]

# Write the workflow report to disk.
report_text = "\n".join(report_lines)
report_path.write_text(report_text)

# Show only a short confirmation summary.
print("Workflow report created successfully.")
print(f"Input file: {input_path.name}")
print(f"Report file: {report_path.name}")
print(f"Average flow: {avg_flow:.1f} gpm")

```

### **3.3. Workflow Overview**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_B/image_03_03.jpg?v=1780930795" width="250">



>* Turn raw data into documented results
>* Repeat calculations consistently across engineering cases

>* Separate input, calculation, and reporting tasks
>* Improve testing, maintenance, reproducibility, and communication

>* Plan reliable paths from data to reports
>* Create repeatable, defensible engineering information



```
#@title Python Code - Workflow Overview

# Demonstrate one automated engineering workflow.
# Read inputs, compute outputs, report results.
# Keep files small and clearly labeled.

from pathlib import Path
import csv

# Define simple files for this workflow.
input_path = Path("beam_input_cases.csv")
report_path = Path("beam_workflow_report.txt")

# Create tiny input data with units.
rows = [
    ["case", "load_lbf", "span_ft", "I_in4"],
    ["A", 450.0, 8.0, 52.0],
    ["B", 600.0, 10.0, 80.0],
]

# Write repeatable input data safely.
with input_path.open("w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)

# Define reusable beam calculation function.
def midspan_deflection(load_lbf, span_ft, inertia_in4):
    modulus_psi = 29000000.0
    span_in = span_ft * 12.0

    # Use simply supported center-load formula.
    deflection_in = load_lbf * span_in**3
    deflection_in /= 48.0 * modulus_psi * inertia_in4
    return deflection_in

# Read engineering inputs from file.
with input_path.open("r", newline="") as file:
    reader = csv.DictReader(file)
    cases = list(reader)

# Validate the small workflow input.
if not cases:
    raise ValueError("No beam cases were found.")

# Compute results for each case.
results = []
for case in cases:
    load = float(case["load_lbf"])
    span = float(case["span_ft"])
    inertia = float(case["I_in4"])

    # Store labeled values for reporting.
    deflection = midspan_deflection(load, span, inertia)
    results.append((case["case"], load, span, inertia, deflection))

# Build a clear engineering report.
report_lines = [
    "Beam Deflection Workflow Report",
    "Assumption: simply supported beam, center point load.",
    "Elastic modulus: 29,000,000 psi for steel.",
    "",
]

# Add one readable line per case.
for name, load, span, inertia, deflection in results:
    line = f"Case {name}: load={load:.1f} lbf, span={span:.1f} ft, "
    line += f"I={inertia:.1f} in^4, deflection={deflection:.5f} in"
    report_lines.append(line)

# Write the final report file.
report_text = "\n".join(report_lines)
report_path.write_text(report_text)

# Print a compact workflow summary.
print("Automated workflow completed successfully.")
print(f"Input file: {input_path.name}")
print(f"Report file: {report_path.name}")
print(f"Cases processed: {len(results)}")
print(report_text)

```
