## **1. Reading Files**

### **1.1. Opening Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_01_01.jpg?v=1780929088" width="250">



>* Open files to access engineering measurements
>* Python connects to located, permitted data

>* Use accurate filenames and file paths
>* Organize folders for repeatable engineering workflows

>* Check file type and data layout
>* Confirm text, units, headers, and line endings



```
#@title Python Code - Opening Files

# Opening files connects programs to measurements.
# This example creates a tiny data file.
# Then Python opens and reads it safely.

# Create a small measurement filename.
file_name = "strain_readings.txt"

# Store simple tensile-test strain readings.
sample_text = "0.0000\n0.0008\n0.0016\n0.0024\n"

# Write the sample file for Colab.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write(sample_text)

# Open the file for reading.
try:
    with open(file_name, "r", encoding="utf-8") as file_object:
        first_line = file_object.readline()

# Report missing files without crashing.
except FileNotFoundError:
    first_line = ""
    print("Measurement file was not found.")

# Convert the first text reading carefully.
try:
    first_strain = float(first_line.strip())

# Use a safe fallback for bad text.
except ValueError:
    first_strain = None

# Print a short engineering result.
if first_strain is not None:
    print("Opened file:", file_name)
    print("First strain reading:", first_strain)
    print("Reading type after conversion:", type(first_strain).__name__)
else:
    print("First reading could not be converted.")

```

### **1.2. Read Mode**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_01_02.jpg?v=1780929090" width="250">



>* Read files without changing original measurements
>* Protect engineering records for later analysis

>* Read whole files or individual lines
>* Line reading suits large measurement records

>* Read file contents as plain text first
>* Parse and verify before engineering decisions



```
#@title Python Code - Read Mode

# Read mode protects original measurement files.
# Engineering text files often contain records.
# We first read text before parsing.

# Create a tiny measurement file safely.
file_name = "shaft_vibration_readings.txt"

# Each line represents one time observation.
sample_text = "time_s,disp_mm\n0.0,0.12\n0.5,0.18\n1.0,0.16\n"

# Write sample data for this demonstration.
with open(file_name, "w", encoding="utf-8") as file:
    file.write(sample_text)

# Open the file using read mode.
with open(file_name, "r", encoding="utf-8") as file:
    first_line = file.readline().strip()
    remaining_text = file.read()

# Show that read mode retrieved text.
print("Header read from file:", first_line)
print("Remaining text block:")
print(remaining_text.strip())

# Reopen and read line by line.
with open(file_name, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Count records without changing the file.
record_count = len(lines) - 1
print("Number of measurement records:", record_count)

```

### **1.3. Closing Files**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_01_03.jpg?v=1780929093" width="250">



>* Close files to release system resources
>* Support reliable engineering data workflows

>* Use automatic cleanup when reading files
>* Prevent open files during unexpected errors

>* Closed files prevent access conflicts and confusion
>* Good habits support reliable, traceable workflows



```
#@title Python Code - Closing Files

# This example demonstrates automatic file closure.
# Engineering measurements are stored as simple text.
# The with statement closes files safely.

# Create a small measurement file.
file_name = "beam_strain_readings.txt"

# Store line based strain readings.
sample_text = "time_s,strain_microstrain\n0.0,105\n0.5,118\n1.0,130\n"

# Write the tiny demonstration file.
with open(file_name, "w", encoding="utf-8") as file_object:
    file_object.write(sample_text)

# Read the file using automatic closure.
with open(file_name, "r", encoding="utf-8") as file_object:
    lines = file_object.readlines()
    closed_inside = file_object.closed

# Check the file state afterward.
closed_after = file_object.closed

# Parse numeric strain values carefully.
strain_values = []
for line in lines[1:]:
    parts = line.strip().split(",")
    strain_values.append(float(parts[1]))

# Calculate a simple engineering summary.
average_strain = sum(strain_values) / len(strain_values)

# Print concise teaching results.
print(f"File closed while reading: {closed_inside}")
print(f"File closed after with block: {closed_after}")
print(f"Read {len(strain_values)} strain measurements.")
print(f"Average strain: {average_strain:.1f} microstrain")

```

## **2. Parsing Numeric Data**

### **2.1. Splitting Data Lines**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_02_01.jpg?v=1780929095" width="250">



>* Separate each text line into measurements
>* Use delimiters before calculating numeric values

>* Match splitting to the actual file format
>* Handle spaces, blanks, units, and artifacts

>* Split pieces prepare data for numeric analysis
>* Test parsing to avoid calculation errors



```
#@title Python Code - Splitting Data Lines

# Parse engineering lines into numeric fields.
# Split text carefully before numeric conversion.
# Handle spaces and missing sensor readings.

# Store small sample data lines.
data_lines = [
    "time_s,pressure_kPa,temperature_C,strain_micro",
    "0.00, 101.3, 22.5, 15",

    "0.10, 104.8, 22.7, 18",
    "0.20, , 23.0, 21",
]

# Select only measured data rows.
measurement_lines = data_lines[1:]
parsed_rows = []

# Split each comma separated line.
for line in measurement_lines:
    pieces = line.split(",")
    cleaned = [piece.strip() for piece in pieces]

    # Convert blank pressure readings safely.
    time_s = float(cleaned[0])
    pressure_kpa = None if cleaned[1] == "" else float(cleaned[1])
    temperature_c = float(cleaned[2])

    strain_micro = float(cleaned[3])
    parsed_rows.append((time_s, pressure_kpa, temperature_c, strain_micro))

# Keep only available pressure values.
valid_pressures = [row[1] for row in parsed_rows if row[1] is not None]
average_pressure = sum(valid_pressures) / len(valid_pressures)

# Show compact parsing results.
print("Original line:", measurement_lines[0])
print("Split pieces:", measurement_lines[0].split(","))
print("Cleaned pieces:", [p.strip() for p in measurement_lines[0].split(",")])
print("Parsed first row:", parsed_rows[0])
print("Valid pressure count:", len(valid_pressures))
print("Average pressure kPa:", round(average_pressure, 2))

```

### **2.2. Splitting Data Lines**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_02_02.jpg?v=1780929097" width="250">



>* Separate each record into individual measurements
>* Prepare text values for numeric calculations

>* Identify delimiters to separate measurement fields
>* Skip headers and notes before calculation

>* Check split fields match expected structure
>* Catch formatting issues before numeric calculations



```
#@title Python Code - Splitting Data Lines

# Split measurement rows into separate fields.
# Check field counts before converting values.
# Keep parsing small and readable.

# Create short comma separated measurement lines.
raw_lines = [
    "time_s,extension_mm,load_N,stress_MPa",
    "0.0,0.00,0,0.0",
    "0.5,0.12,850,42.5",

    "1.0,0.25,1710,85.5",
    "note,operator checked grips",
    "1.5,0.39,2520,126.0",
]

# Define the expected measurement field count.
expected_fields = 4

# Store accepted numeric measurement records.
records = []

# Parse each line using comma splitting.
for line_number, line_text in enumerate(raw_lines, start=1):
    pieces = line_text.split(",")

    # Skip header and note lines safely.
    if pieces[0] in ["time_s", "note"]:
        continue

    # Reject rows with the wrong length.
    if len(pieces) != expected_fields:
        print("Skipped line", line_number, "field count", len(pieces))
        continue

    # Convert split text fields into floats.
    try:
        values = [float(piece) for piece in pieces]
    except ValueError:
        print("Skipped line", line_number, "conversion problem")
        continue

    # Save only complete numeric records.
    records.append(values)

# Calculate a simple engineering summary.
peak_load = max(row[2] for row in records)
final_stress = records[-1][3]

# Show compact parsing results.
print("Valid numeric rows:", len(records))
print("First parsed row:", records[0])
print("Peak load N:", peak_load)
print("Final stress MPa:", final_stress)

```

### **2.3. Float Conversion**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_02_03.jpg?v=1780929099" width="250">



>* Turn file text into usable numbers
>* Enable calculations with engineering measurements

>* Floats handle decimal engineering measurements
>* Clean text before converting values

>* Check data quality before converting floats
>* Handle invalid values transparently and safely



```
#@title Python Code - Float Conversion

# Text file values arrive as strings.
# Float conversion enables engineering calculations.
# Bad fields need controlled handling.

# Create tiny engineering file lines.
file_lines = [
    "time_s,force_N,deflection_mm",
    "0.00,125.5,0.012",

    "0.10,130.2,0.014",
    "0.20,missing,0.017",
    "0.30,142.8,0.019",
]

# Store sample data in one file.
file_name = "beam_test_readings.csv"
with open(file_name, "w") as output_file:
    output_file.write("\n".join(file_lines))

# Read the file as text.
valid_forces = []
problems = []
with open(file_name, "r") as input_file:
    rows = input_file.readlines()

# Skip header and parse measurements.
for row_number, row_text in enumerate(rows[1:], start=2):
    fields = row_text.strip().split(",")
    try:
        force_value = float(fields[1])

        valid_forces.append(force_value)
    except ValueError:
        problems.append(row_number)

# Check valid data before averaging.
if valid_forces:
    average_force = sum(valid_forces) / len(valid_forces)
else:
    average_force = float("nan")

# Print concise engineering results.
print("Converted force readings:", valid_forces)
print("Rows with conversion problems:", problems)
print("Average valid force, N:", round(average_force, 2))

```

## **3. File Error Handling**

### **3.1. File Not Found**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_03_01.jpg?v=1780929103" width="250">



>* Missing or incorrect paths stop file access
>* Expect this when using external data

>* Missing files often reflect workflow issues
>* Clear messages identify paths and causes

>* Stop analysis when required files are missing
>* Use organized files for reliable calculations



```
#@title Python Code - File Not Found

# Demonstrate controlled file missing handling.
# Use a small engineering data filename.
# Keep messages clear for troubleshooting.

# Define the expected measurement file path.
file_path = "project_data/strain_log_test01.csv"

# Prepare variables for controlled workflow status.
file_loaded = False
measurement_count = 0

# Try opening the required engineering file.
try:
    with open(file_path, "r", encoding="utf-8") as data_file:

        # Read only a small preview safely.
        preview_lines = data_file.readlines()[:3]
        measurement_count = len(preview_lines)

        # Mark the workflow as successful.
        file_loaded = True

# Handle the missing file clearly.
except FileNotFoundError:
    print("File not found during engineering data loading.")
    print(f"Expected file path: {file_path}")

    # Explain the controlled workflow response.
    print("Analysis stopped until the measurement file is available.")
    print("Check folder name, filename, extension, and location.")

# Report a safe final workflow status.
if file_loaded:
    print(f"Loaded preview lines: {measurement_count}")
else:
    print("No calculations were performed with missing data.")

```

### **3.2. Value Conversion Errors**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_03_02.jpg?v=1780929101" width="250">



>* Text measurements must become usable numbers
>* Messy entries can break calculations

>* Treat conversion problems as data quality issues
>* Log bad values; skip or stop safely

>* Validate varied file formats before calculation
>* Report errors to ensure reliable results



```
#@title Python Code - Value Conversion Errors

# Demonstrate value conversion errors safely.
# Read a tiny engineering measurement file.
# Report rejected rows without stopping.

# Create a small comma separated sample file.
file_name = "load_test_measurements.csv"

# Include valid and flawed engineering records.
sample_lines = [
    "time_s,force_kN,strain_mm_per_mm",
    "0.0,0.0,0.0000",
    "1.0,12.4 kN,0.0012",

    "2.0,18.6,unavailable",
    "3.0,22.1,0.0024",
]

# Write the sample data using builtins.
with open(file_name, "w", encoding="utf-8") as file_handle:
    file_handle.write("\n".join(sample_lines))

# Prepare containers for accepted and rejected rows.
valid_forces = []
rejected_rows = []

# Open the file with controlled error handling.
try:
    with open(file_name, "r", encoding="utf-8") as file_handle:
        rows = file_handle.readlines()
except FileNotFoundError:
    rows = []

# Convert each data row carefully.
for line_number, row_text in enumerate(rows[1:], start=2):
    cells = row_text.strip().split(",")

    # Reject rows that do not match expected columns.
    if len(cells) != 3:
        rejected_rows.append((line_number, "wrong column count"))
        continue

    # Try converting text values into numbers.
    try:
        time_s = float(cells[0])
        force_kN = float(cells[1])
        strain = float(cells[2])

    # Store context when conversion fails.
    except ValueError:
        rejected_rows.append((line_number, row_text.strip()))
        continue

    # Keep only validated numeric force values.
    valid_forces.append(force_kN)

# Compute a result only from trusted values.
average_force = sum(valid_forces) / len(valid_forces)

# Print a compact engineering quality summary.
print("Valid force readings:", len(valid_forces))
print("Rejected rows:", len(rejected_rows))
print("Average valid force kN:", round(average_force, 2))
print("First rejection:", rejected_rows[0])

```

### **3.3. Safe File Handling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_04/Lecture_A/image_03_03.jpg?v=1780929105" width="250">



>* Open, use, and close files carefully
>* Prevent locked files and workflow conflicts

>* Clean up files after errors
>* Prevent locks and workflow disruptions

>* Clearly report file reading failures
>* Protect original data for trusted calculations



```
#@title Python Code - Safe File Handling

# Safely read small engineering measurement files.
# Handle missing files and conversion problems.
# Keep file access controlled and clear.

from pathlib import Path

# Create a small sample measurement file.
file_path = Path("pressure_readings_psi.csv")

# Include one bad row for demonstration.
sample_text = "time_s,pressure_psi\n0.0,14.7\n0.5,15.1\n1.0,bad_value\n1.5,15.6\n"

# Write the tiny file safely.
file_path.write_text(sample_text, encoding="utf-8")

# Prepare containers for valid values.
pressures = []
problem_lines = []

# Read the file using context management.
try:
    with file_path.open("r", encoding="utf-8") as file:
        header = file.readline().strip()

        # Process each remaining data line.
        for line_number, line in enumerate(file, start=2):
            parts = line.strip().split(",")

            # Validate the expected comma separated fields.
            if len(parts) != 2:
                problem_lines.append(line_number)

            else:
                try:
                    pressure = float(parts[1])
                    pressures.append(pressure)
                except ValueError:
                    problem_lines.append(line_number)

# Report controlled file access problems.
except FileNotFoundError:
    print("Engineering file was not found.")

except OSError as error:
    print(f"File system problem: {error}")

# Summarize successful conversion results.
if pressures:
    average_pressure = sum(pressures) / len(pressures)
    print(f"Header read: {header}")
    print(f"Valid pressure readings: {len(pressures)}")

    print(f"Average pressure: {average_pressure:.2f} psi")
    print(f"Problem lines skipped: {problem_lines}")
else:
    print("No valid pressure readings were converted.")

```
