## **1. Dictionary Basics**

### **1.1. Key Value Basics**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_01_01.jpg?v=1780927426" width="250">



>* Keys label stored engineering values
>* Dictionaries make properties easier to retrieve

>* Choose clear keys for engineering properties
>* Retrieve values by name to avoid errors

>* Shared keys organize different engineering cases
>* Consistent names simplify comparison and reuse



```
#@title Python Code - Key Value Basics

# Dictionaries connect names to engineering values.
# Keys retrieve values without remembering positions.
# This example stores material properties clearly.

# Create one dictionary for an alloy.
material = {
    "name": "6061-T6 aluminum",

    "density_kg_m3": 2700,
    "elastic_modulus_GPa": 68.9,
    "yield_strength_MPa": 276,

    "typical_use": "machined bracket",
}

# Retrieve values by descriptive keys.
name_text = material["name"]
density_value = material["density_kg_m3"]
modulus_value = material["elastic_modulus_GPa"]
yield_value = material["yield_strength_MPa"]

# Build readable labels from strings.
property_label = "Yield strength".upper()
units_label = "MPa"

# Use get for a safe missing lookup.
poisson_value = material.get("poisson_ratio", "not recorded")
use_text = material.get("typical_use", "general component")

# Print a compact engineering report.
print("Material dictionary lookup report")
print(f"Material: {name_text}")
print(f"Density: {density_value} kg/m^3")
print(f"Elastic modulus: {modulus_value} GPa")
print(f"{property_label}: {yield_value} {units_label}")
print(f"Poisson ratio: {poisson_value}")
print(f"Typical use: {use_text}")
print("Keys available:", ", ".join(material.keys()))

```

### **1.2. Material Property Lookup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_01_02.jpg?v=1780927428" width="250">



>* Material names retrieve related engineering properties
>* Named lookups make code clearer

>* Compare materials with repeatable property lookups
>* Keep data separate from engineering calculations

>* Use specific names for material properties
>* Treat values as simplified engineering assumptions



```
#@title Python Code - Material Property Lookup

# Dictionaries connect names to engineering property values.
# Material lookup avoids mysterious numeric index positions.
# Clear labels improve readable mechanical engineering calculations.

# Store simplified material properties in nested dictionaries.
material_properties = {
    "aluminum 6061": {
        "density_kg_m3": 2700,

        "elastic_modulus_gpa": 69,
        "yield_strength_mpa": 276,
    },

    "structural steel": {
        "density_kg_m3": 7850,
        "elastic_modulus_gpa": 200,

        "yield_strength_mpa": 250,
    },
}

# Choose one material name for lookup.
selected_material = "aluminum 6061"

# Retrieve the selected material dictionary safely.
selected_properties = material_properties.get(selected_material, {})

# Retrieve named properties from the selected material.
density = selected_properties.get("density_kg_m3")
yield_strength = selected_properties.get("yield_strength_mpa")

# Define bracket dimensions using metric units.
volume_cm3 = 120
volume_m3 = volume_cm3 / 1_000_000

# Compute mass from density and volume.
mass_kg = density * volume_m3

# Build readable labels using formatted strings.
material_label = selected_material.title()
property_label = "Yield strength"

# Print a compact material property report.
print(f"Material: {material_label}")
print(f"Density: {density} kg/m^3")
print(f"{property_label}: {yield_strength} MPa")
print(f"Bracket volume: {volume_cm3} cm^3")
print(f"Estimated mass: {mass_kg:.3f} kg")

```

### **1.3. Units and Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_01_03.jpg?v=1780927430" width="250">



>* Engineering values need units for meaning
>* Dictionaries link properties to usable data

>* Plan consistent units before storing values
>* Prevent hidden conversion errors in calculations

>* Store units with values for clarity
>* Check meaning, measurement method, and context



```
#@title Python Code - Units and Values

# Dictionaries can store units with engineering values.
# Clear labels help prevent unit mistakes.
# This example keeps units beside numbers.

# Store material properties with explicit units.
aluminum = {
    "name": "Aluminum 6061-T6",
    "density": {"value": 2700, "unit": "kg/m^3"},
    "yield_strength": {"value": 276, "unit": "MPa"},
}

# Store a geometry value using imperial units.
shaft = {
    "diameter": {"value": 2.0, "unit": "in"},
    "length": {"value": 18.0, "unit": "in"},
}

# Retrieve values and units by name.
density_value = aluminum["density"]["value"]
density_unit = aluminum["density"]["unit"]
yield_value = aluminum["yield_strength"]["value"]
yield_unit = aluminum["yield_strength"]["unit"]

# Build readable engineering labels with strings.
density_label = f"Density: {density_value} {density_unit}"
yield_label = f"Yield strength: {yield_value} {yield_unit}"
shaft_label = f"Shaft diameter: {shaft['diameter']['value']} {shaft['diameter']['unit']}"

# Print a short report for the learner.
print("Material property report")
print(f"Material: {aluminum['name']}")
print(density_label)
print(yield_label)
print(shaft_label)

# Show why units belong with stored values.
print("Always read the unit before calculating.")

```

## **2. Structured Case Calculations**

### **2.1. Material Function Lookup**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_02_01.jpg?v=1780927418" width="250">



>* Map material names to needed calculations
>* Avoid long choices and improve program structure

>* Different materials need different calculation methods
>* Dictionary lookup selects the correct function

>* Use clear names, units, and assumptions
>* Add materials without rewriting program logic



```
#@title Python Code - Material Function Lookup

# Dictionaries can map materials to calculation functions.
# This example estimates tensile elongation for selected cases.
# Function lookup avoids long material choice chains.

# Store modulus values using consistent MPa units.
material_data = {
    "steel": {"E_MPa": 200000.0},
    "aluminum": {"E_MPa": 69000.0},
    "polymer": {"E_MPa": 3000.0},
}

# Define a linear elastic strain calculation function.
def metal_elongation(force_N, area_mm2, length_mm, E_MPa):
    stress_MPa = force_N / area_mm2
    strain = stress_MPa / E_MPa

    elongation_mm = strain * length_mm
    return elongation_mm

# Define a polymer function with simple safety adjustment.
def polymer_elongation(force_N, area_mm2, length_mm, E_MPa):
    stress_MPa = force_N / area_mm2
    strain = stress_MPa / E_MPa

    creep_factor = 1.35
    elongation_mm = strain * length_mm * creep_factor
    return elongation_mm

# Map material names directly to calculation functions.
material_functions = {
    "steel": metal_elongation,
    "aluminum": metal_elongation,
    "polymer": polymer_elongation,
}

# Define one small tensile design case.
case = {"force_N": 12000.0, "area_mm2": 150.0, "length_mm": 500.0}
selected_materials = ["steel", "aluminum", "polymer"]

# Check each selected material before calculating results.
results = []
for material in selected_materials:
    if material not in material_data:
        continue

    calculator = material_functions.get(material)
    properties = material_data[material]
    elongation = calculator(**case, **properties)

    label = material.title() + " tensile member"
    results.append((label, elongation))

# Print a compact engineering report.
print("Material function lookup report")
print("Case: 12000 N load, 150 mm^2 area, 500 mm length")
for label, elongation in results:
    print(f"{label}: elongation = {elongation:.3f} mm")

```

### **2.2. Selecting Case Parameters**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_02_02.jpg?v=1780927413" width="250">



>* Dictionaries define specific engineering cases
>* Named inputs make functions clearer and consistent

>* Compare design options using shared parameter names
>* Reuse one function for different cases

>* Named parameters reduce mistakes and confusion
>* Structured cases support reports and review



```
#@title Python Code - Selecting Case Parameters

# Dictionaries can organize engineering case parameters.
# Functions can calculate selected case results.
# Strings can label reports clearly.

# Store material properties in named dictionaries.
materials = {
    "Aluminum 6061-T6": {
        "density_kg_m3": 2700,

        "yield_strength_mpa": 276,
    },
    "Mild steel": {
        "density_kg_m3": 7850,

        "yield_strength_mpa": 250,
    },
}

# Store one selected design case.
case = {
    "case_name": "robot arm link",
    "material": "Aluminum 6061-T6",

    "length_m": 0.80,
    "area_cm2": 3.0,
    "safety_factor": 2.0,
}

# Define a function using named entries.
def evaluate_link(selected_case, material_table):
    material_name = selected_case["material"]
    material_data = material_table[material_name]

    area_m2 = selected_case["area_cm2"] / 10000
    volume_m3 = selected_case["length_m"] * area_m2
    mass_kg = volume_m3 * material_data["density_kg_m3"]

    allowable_mpa = material_data["yield_strength_mpa"] / selected_case["safety_factor"]
    return mass_kg, allowable_mpa

# Compute results for the selected case.
mass_kg, allowable_mpa = evaluate_link(case, materials)
label = case["case_name"].title()

# Print a compact engineering report.
print(f"Case: {label}")
print(f"Material: {case['material']}")
print(f"Length: {case['length_m']} m")
print(f"Area: {case['area_cm2']} cm^2")
print(f"Estimated mass: {mass_kg:.2f} kg")
print(f"Allowable stress: {allowable_mpa:.1f} MPa")

```

### **2.3. Material Case Functions**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_02_03.jpg?v=1780927416" width="250">



>* Store material properties in dictionaries
>* Use functions to calculate selected cases

>* Reuse functions across compatible material entries
>* Compare cases by changing dictionary selections

>* Show required material properties clearly
>* Link results to traceable case labels



```
#@title Python Code - Material Case Functions

# Material dictionaries organize engineering property values.
# Functions reuse calculations for selected cases.
# String labels make results easier to read.

# Store material properties in consistent dictionaries.
materials = {
    "Aluminum 6061": {"E_GPa": 69.0, "density": 2700},
    "Structural Steel": {"E_GPa": 200.0, "density": 7850},
    "Titanium Ti-6Al-4V": {"E_GPa": 114.0, "density": 4430},
}

# Store the shared beam loading case.
case = {"name": "tip load case", "P_N": 500.0,
        "L_m": 1.2, "I_m4": 2.5e-6}

# Define one reusable material calculation function.
def beam_tip_deflection_mm(material_name, material_data, load_case):
    required_keys = ["E_GPa", "density"]
    missing_keys = [key for key in required_keys if key not in material_data]

    if missing_keys:
        missing_text = ", ".join(missing_keys)
        return f"{material_name}: missing {missing_text}"

    elastic_modulus_pa = material_data["E_GPa"] * 1e9
    deflection_m = load_case["P_N"] * load_case["L_m"] ** 3

    deflection_m /= 3.0 * elastic_modulus_pa * load_case["I_m4"]
    deflection_mm = deflection_m * 1000.0
    label = material_name.upper()

    return f"{label}: {deflection_mm:.3f} mm deflection"

# Print a compact report for each material.
print("Structured beam case report")
print(f"Case label: {case['name'].title()}")

# Apply the same function to every material.
for name, data in materials.items():
    result = beam_tip_deflection_mm(name, data, case)
    print(result)

# Parse a simple engineering input string.
input_text = "material=Aluminum 6061; case=tip load case"
selected = input_text.split(";")[0].split("=")[1].strip()

# Reuse the function for the parsed material.
selected_result = beam_tip_deflection_mm(selected, materials[selected], case)
print(f"Parsed selection -> {selected_result}")

```

```
#@title Python Code - Engineering Labels

# Engineering labels clarify numerical mechanical data.
# Strings can combine properties and units.
# Consistent labels reduce ambiguity in reports.

# Store compact engineering properties in dictionaries.
material = {"name": "stainless_steel", "density": 7850, "unit": "kg/m^3"}
case = {"specimen": "shaft", "diameter": 25, "diameter_unit": "mm"}

# Convert internal material names into readable text.
material_name = material["name"].replace("_", " ").title()
specimen_name = case["specimen"].title()

# Build one clear engineering label.
label_parts = [material_name, specimen_name, "Density Check"]
engineering_label = " | ".join(label_parts)

# Format numerical data with units.
density_text = f"Density = {material['density']} {material['unit']}"
diameter_text = f"Diameter = {case['diameter']} {case['diameter_unit']}"

# Parse a compact engineering input string.
raw_input = "beam: aluminum_6061 : length=48 in"
piece, alloy, measurement = [part.strip() for part in raw_input.split(":")]

# Standardize parsed text for reporting.
alloy_label = alloy.replace("_", " ").title()
parsed_label = f"{piece.title()} | {alloy_label} | {measurement}"

# Print a short readable engineering report.
print("ENGINEERING LABEL EXAMPLES")
print(engineering_label)
print(density_text)
print(diameter_text)
print(parsed_label)

```

### **3.2. String Formatting**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_03_02.jpg?v=1780927424" width="250">



>* Format values with labels, units, and context
>* Clear text prevents engineering misunderstandings

>* Control numeric precision for engineering decisions
>* Create consistent, readable reports for review

>* Show context, assumptions, and result status
>* Use clear units, labels, and rounding



```
#@title Python Code - String Formatting

# Format engineering values clearly.
# Labels make calculations understandable.
# F-strings control readable output.

# Store named material properties.
materials = {
    "steel": {"density": 7850, "yield_mpa": 250},
    "aluminum": {"density": 2700, "yield_mpa": 95},
}

# Store one load case.
case = {
    "name": "tie rod A",
    "material": "steel",
    "force_n": 12000,
    "area_mm2": 80,
}

# Compute axial stress safely.
stress_mpa = case["force_n"] / case["area_mm2"]

# Retrieve the selected material.
material = materials[case["material"]]

# Compute factor of safety.
factor_safety = material["yield_mpa"] / stress_mpa

# Build a readable engineering label.
label = f"{case['name'].title()} using {case['material'].title()}"

# Format values with units.
stress_text = f"Axial stress: {stress_mpa:.1f} MPa"
fos_text = f"Factor of safety: {factor_safety:.2f}"

# Format density with thousands separator.
density_text = f"Density: {material['density']:,} kg/m^3"

# Parse a compact imperial input.
imperial_input = "length=24 in"
parts = imperial_input.replace("=", " ").split()

# Convert parsed length to centimeters.
length_cm = float(parts[1]) * 2.54

# Report parsed and formatted length.
length_text = f"Parsed length: {length_cm:.1f} cm"

# Print a short formatted report.
print("Engineering string formatting report")
print(f"Component: {label}")
print(density_text)
print(stress_text)
print(fos_text)
print(length_text)

```

### **3.3. Parsed Engineering Inputs**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_B/image_03_03.jpg?v=1780927421" width="250">



>* Turn engineering text into usable data
>* Extract properties, values, units, and labels

>* Treat typed text as useful but unreliable
>* Normalize inputs for dependable engineering calculations

>* Parse labels into reusable engineering data
>* Flag unclear inputs for reliable workflows



```
#@title Python Code - Parsed Engineering Inputs

# Parse text inputs into engineering fields.
# Normalize units before using values.
# Report structured results for review.

# Store rough entries typed by different people.
raw_entries = [
    "material: steel; diameter = 10 mm; length = 2 m",
    " material: Aluminum ; diameter=0.50 in; length=24 in ",
    "material: stainless-steel; diameter = 12 millimeters; length=1.5 meters",
]

# Define accepted unit aliases for normalization.
unit_aliases = {
    "millimeters": "mm", "millimetres": "mm", "mm": "mm",
    "meters": "m", "metres": "m", "m": "m",
    "inches": "in", "inch": "in", "in": "in",
}

# Define consistent material names for reports.
material_aliases = {
    "steel": "Steel",
    "aluminum": "Aluminum",
    "aluminium": "Aluminum",
    "stainless-steel": "Stainless steel",
}

# Parse one name, value, and unit phrase.
def parse_quantity(text_piece):
    name_part, value_part = text_piece.split("=", 1)
    words = value_part.strip().lower().split()
    value = float(words[0]) if words else None

    unit = unit_aliases.get(words[1], "unknown") if len(words) > 1 else "unknown"
    return name_part.strip().lower(), value, unit

# Parse each rough engineering description.
parsed_cases = []
for entry in raw_entries:
    pieces = [piece.strip() for piece in entry.split(";")]
    material_key = pieces[0].split(":", 1)[1].strip().lower()

    case = {"material": material_aliases.get(material_key, "Review needed")}
    for piece in pieces[1:]:
        name, value, unit = parse_quantity(piece)
        case[name] = {"value": value, "unit": unit}

    parsed_cases.append(case)

# Print a compact engineering parsing report.
print("Parsed engineering inputs:")
for index, case in enumerate(parsed_cases, start=1):
    diameter = case["diameter"]
    length = case["length"]
    label = f"Case {index}: {case['material']}"

    result = f"{label}, diameter {diameter['value']} {diameter['unit']}, length {length['value']} {length['unit']}"
    print(result)

```
