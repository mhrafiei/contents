## **1. List Operations**

### **1.1. Indexing Measurement Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_01_01.jpg?v=1780926442" width="250">



>* Indexes locate measurements by list position
>* Python starts counting positions at zero

>* Index shows a reading’s list position
>* Use measurement context to avoid misreading

>* Wrong indexes can mislead engineering decisions
>* Check zero-based and negative positions carefully



```
#@title Python Code - Indexing Measurement Lists

# Store measurements in an ordered Python list.
# Indices select values by their positions.
# Negative indices count backward from endings.

# Define chronological pressure measurements in kilopascals.
pressure_kpa = [101.2, 103.5, 106.1, 104.8, 108.3]

# Check the list length before indexing.
measurement_count = len(pressure_kpa)
assert measurement_count == 5, "Unexpected measurement count"

# Access selected readings using zero-based indices.
first_reading = pressure_kpa[0]
third_reading = pressure_kpa[2]
final_reading = pressure_kpa[-1]

# Convert an engineering measurement number into index.
tenth_measurement_number = 4
tenth_measurement_index = tenth_measurement_number - 1
selected_reading = pressure_kpa[tenth_measurement_index]

# Print concise results for learner inspection.
print("Pressure readings, kPa:", pressure_kpa)
print("First reading, index 0:", first_reading)
print("Third reading, index 2:", third_reading)
print("Final reading, index -1:", final_reading)
print("Measurement 4 uses index 3:", selected_reading)

```

### **1.2. Slicing Lists**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_01_02.jpg?v=1780926445" width="250">



>* Select ordered measurement segments from lists
>* Study patterns within important data intervals

>* Choose slice start and stop positions carefully
>* Precise ranges preserve correct data interpretation

>* Analyze subsets while preserving original measurement lists
>* Use ordered slices for focused engineering evidence



```
#@title Python Code - Slicing Lists

# Store ordered pump temperatures in a list.
# Slice useful operating intervals from measurements.
# Compare original data with selected subsets.

# Create temperature readings in degrees Celsius.
temperatures_c = [68, 72, 75, 79, 82, 84, 83, 81]

# Select readings from warm-up minutes two through five.
warmup_slice = temperatures_c[1:5]

# Select the final three recorded readings.
final_slice = temperatures_c[-3:]

# Select every second reading for a quick review.
every_second_reading = temperatures_c[::2]

# Print the original ordered measurement list.
print("All readings:", temperatures_c)

# Print the warm-up slice result.
print("Warm-up slice:", warmup_slice)

# Print the final slice result.
print("Final three readings:", final_slice)

# Print the skipped interval slice result.
print("Every second reading:", every_second_reading)

# Confirm slicing did not change original data.
print("Original unchanged:", temperatures_c)

```

### **1.3. Accessing List Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_01_03.jpg?v=1780926447" width="250">



>* List order gives measurements their meaning
>* Use zero-based positions to access values

>* Choose correct positions for reliable engineering decisions
>* Check list size to avoid invalid access

>* Use list values to inspect measurements
>* Keep names and ordering physically meaningful



```
#@title Python Code - Accessing List Values

# Lists store ordered engineering measurements.
# Indexing retrieves one selected measurement.
# Python list indexes start at zero.

# Store coolant temperatures measured over time.
temperatures_c = [24.8, 31.2, 38.5, 44.1, 47.0]

# Count values before accessing positions.
reading_count = len(temperatures_c)

# Access first, third, and final readings.
initial_temp = temperatures_c[0]
third_temp = temperatures_c[2]
final_temp = temperatures_c[-1]

# Compare selected readings for interpretation.
temperature_rise = final_temp - initial_temp

# Print values with engineering meaning.
print("Coolant readings stored:", reading_count)
print("Initial temperature (C):", initial_temp)
print("Third temperature (C):", third_temp)
print("Final temperature (C):", final_temp)
print("Temperature rise (C):", round(temperature_rise, 1))

# Access a safe index only when valid.
requested_index = 3

# Check list length before indexing.
if requested_index < reading_count:
    selected_temp = temperatures_c[requested_index]
    print("Reading at index 3 (C):", selected_temp)
else:
    print("Requested index is outside the list.")

```

## **2. Data Summaries**

### **2.1. Mean Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_02_01.jpg?v=1780926451" width="250">



>* Mean summarizes repeated engineering measurements
>* Add values, then divide by count

>* Loops total changing engineering measurement lists
>* Built-ins quickly compute typical operating values

>* Means can hide outliers or unusual events
>* Check data quality before interpreting means



```
#@title Python Code - Mean Values

# Mean values summarize repeated engineering measurements.
# Lists store related test readings together.
# Builtins compute totals and counts safely.

# No extra package installation is needed.

# Store tensile strengths in megapascals.
tensile_strengths_mpa = [412, 418, 409, 421, 415]

# Start a running total manually.
running_total = 0

# Add each measurement during looping.
for strength in tensile_strengths_mpa:
    running_total = running_total + strength

# Count readings before dividing values.
reading_count = len(tensile_strengths_mpa)

# Compute the loop based mean.
loop_mean = running_total / reading_count

# Compute the same mean using builtins.
builtin_mean = sum(tensile_strengths_mpa) / len(tensile_strengths_mpa)

# Print concise engineering summary results.
print("Tensile strengths MPa:", tensile_strengths_mpa)
print("Number of readings:", reading_count)
print("Loop mean MPa:", round(loop_mean, 1))
print("Builtin mean MPa:", round(builtin_mean, 1))
print("Means match:", loop_mean == builtin_mean)

```

### **2.2. Average Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_02_02.jpg?v=1780926449" width="250">



>* Averages show typical engineering measurement values
>* Functions and loops help calculate summaries

>* Average only comparable engineering measurements
>* Filter data using engineering context

>* Averages can hide variation and extremes
>* Use summaries and loops for context



```
#@title Python Code - Average Values

# Average values summarize repeated measurements.
# Lists store comparable engineering readings.
# Built-ins make summaries clear and reliable.

# Store repeated pump flow measurements.
flow_rates_lpm = [49.8, 50.4, 51.1, 49.5, 50.2]

# Count values before calculating an average.
reading_count = len(flow_rates_lpm)

# Add all list values using sum.
total_flow_lpm = sum(flow_rates_lpm)

# Compute the average only when data exist.
average_flow_lpm = total_flow_lpm / reading_count

# Store a fixed label and unit pair.
flow_label = ("Pump flow", "L/min")

# Use a loop for filtered averaging.
stable_flows_lpm = []

# Keep readings after startup stabilization.
for flow_value in flow_rates_lpm[1:]:
    stable_flows_lpm.append(flow_value)

# Summarize the filtered stable readings.
stable_average_lpm = sum(stable_flows_lpm) / len(stable_flows_lpm)

# Print concise engineering summary results.
print("Measurement:", flow_label[0], "in", flow_label[1])
print("All readings:", flow_rates_lpm)
print("Average flow:", round(average_flow_lpm, 2), flow_label[1])
print("Stable average:", round(stable_average_lpm, 2), flow_label[1])

```

### **2.3. Range Summary**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_02_03.jpg?v=1780926453" width="250">



>* Range shows smallest-to-largest measurement spread
>* Lists and functions quickly find extremes

>* Loop through data to track extremes
>* Range quickly shows measurement fluctuation

>* Interpret range using engineering context
>* Compare with averages to spot instability



```
#@title Python Code - Range Summary

# Range summaries reveal measurement spread.
# Lists can store operating readings.
# Loops can find extreme values.

# Store pump pressure readings in psi.
pressure_readings_psi = [48.2, 49.1, 47.8, 50.0, 48.9, 46.7]

# Validate the list before summarizing.
if len(pressure_readings_psi) == 0:
    raise ValueError("Pressure list must contain measurements")

# Start loop summary using first reading.
lowest_pressure = pressure_readings_psi[0]
highest_pressure = pressure_readings_psi[0]

# Compare each reading with saved extremes.
for pressure_value in pressure_readings_psi:
    if pressure_value < lowest_pressure:
        lowest_pressure = pressure_value

    if pressure_value > highest_pressure:
        highest_pressure = pressure_value

# Compute range after loop comparisons.
pressure_range = highest_pressure - lowest_pressure

# Compute same summary using built-ins.
builtin_low = min(pressure_readings_psi)
builtin_high = max(pressure_readings_psi)
builtin_range = builtin_high - builtin_low

# Store fixed acceptable limits as tuple.
acceptable_window_psi = (45.0, 52.0)

# Check whether extremes meet limits.
within_limits = (
    builtin_low >= acceptable_window_psi[0]
    and builtin_high <= acceptable_window_psi[1]
)

# Print a compact engineering summary.
print("Pump pressure readings:", pressure_readings_psi)
print("Loop summary low/high/range:", lowest_pressure, highest_pressure, pressure_range)
print("Built-in summary low/high/range:", builtin_low, builtin_high, builtin_range)
print("Acceptable window psi:", acceptable_window_psi)
print("All observed pressures within limits:", within_limits)

```

## **3. Tuple Uses**

### **3.1. Tuple Fundamentals**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_03_01.jpg?v=1780926436" width="250">



>* Tuples store ordered values in fixed groups
>* Positions give each engineering value meaning

>* Tuples signal values should stay unchanged
>* Use tuples for small stable records

>* Keep tuple positions clear and meaningful
>* Use tuples for small related fixed groups



```
#@title Python Code - Tuple Fundamentals

# Tuples group related engineering values.
# Tuple positions carry engineering meaning.
# Fixed records improve program clarity.

# Store one fixed sensor location tuple.
sensor_location_m = (2.5, 1.2, 0.8)

# Unpack tuple values into named variables.
x_m, y_m, z_m = sensor_location_m

# Display tuple and accessed components.
print("Sensor location tuple:", sensor_location_m)
print("x coordinate in meters:", x_m)
print("elevation coordinate in meters:", z_m)

# Store fixed bearing dimensions using inches.
bearing_dimensions_in = (1.25, 2.75, 0.65)

# Unpack dimensions for clearer calculations.
inner_diameter_in, outer_diameter_in, width_in = bearing_dimensions_in

# Calculate radial wall thickness safely.
wall_thickness_in = (outer_diameter_in - inner_diameter_in) / 2

# Print engineering meaning for each position.
print("Bearing dimensions tuple:", bearing_dimensions_in)
print("inner, outer, width inches:", bearing_dimensions_in)
print("radial wall thickness inches:", round(wall_thickness_in, 3))

# Show tuples reject item reassignment.
try:
    sensor_location_m[0] = 3.0
except TypeError as error:
    print("Tuples cannot be edited:", type(error).__name__)

# Confirm tuple is still unchanged.
print("Unchanged sensor location:", sensor_location_m)

```

### **3.2. Fixed Value Groups**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_03_02.jpg?v=1780926440" width="250">



>* Tuples group related engineering values together
>* Use tuples for fixed, stable data

>* Tuples group related engineering attributes together
>* Element order needs clear documentation

>* Tuples help protect stable engineering values
>* Changes become deliberate, clear, and safer



```
#@title Python Code - Fixed Value Groups

# Tuples group engineering values together.
# Fixed groups should not change casually.
# Lists suit data that may grow.

# Store one fixed coordinate tuple.
hole_center_mm = (35.0, 18.0)

# Unpack positions using the agreed order.
x_mm, y_mm = hole_center_mm

# Store one fixed material property tuple.
steel_properties = (7850, 200e9, 250e6)

# Unpack density, modulus, and yield strength.
density, elastic_modulus, yield_strength = steel_properties

# Store one fixed imperial bolt description.
bolt_spec_in = (0.25, 20, 1.50)

# Print the grouped values clearly.
print("Hole center tuple:", hole_center_mm)
print("x position in mm:", x_mm)
print("y position in mm:", y_mm)

# Print selected material properties.
print("Steel density kg/m^3:", density)
print("Elastic modulus GPa:", elastic_modulus / 1e9)
print("Yield strength MPa:", yield_strength / 1e6)

# Show tuple length confirms fixed structure.
print("Bolt tuple item count:", len(bolt_spec_in))
print("Bolt diameter in:", bolt_spec_in[0])

# Reassigning creates a new tuple deliberately.
updated_center_mm = (36.0, 18.0)
print("Updated center tuple:", updated_center_mm)

```

### **3.3. Tuple Return Values**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/LFF/Python for Mechanical Engineers/Module_03/Lecture_A/image_03_03.jpg?v=1780926438" width="250">



>* Return related results as one tuple
>* Keep fixed values ordered and connected

>* Know each tuple position’s meaning
>* Group fixed, connected engineering results

>* Use tuples for small, clear grouped results
>* Choose descriptive structures for confusing many-field outputs



```
#@title Python Code - Tuple Return Values

# Tuples can bundle fixed engineering results.
# Functions can return several related values.
# Unpacking assigns each returned position clearly.

# Define a small tensile-test summary function.
def summarize_tensile_test(loads_kn, area_mm2):

    # Validate equal measurement collection sizes.
    if len(loads_kn) == 0:

        # Return a fixed three-value result.
        return (0.0, 0.0, "no data")

    # Convert kilonewtons and millimeters into megapascals.
    stresses_mpa = [load * 1000 / area_mm2 for load in loads_kn]

    # Select simple engineering summary values.
    ultimate_mpa = max(stresses_mpa)
    first_mpa = stresses_mpa[0]

    # Create a compact diagnostic status.
    status = "passes" if ultimate_mpa >= 250 else "review"

    # Return related outputs as one tuple.
    return (ultimate_mpa, first_mpa, status)

# Store repeated load measurements in a list.
tensile_loads_kn = [12.0, 18.5, 24.0, 22.0]

# Store fixed specimen geometry as a tuple.
specimen_geometry = (100.0, 95.0)

# Unpack tuple geometry into named values.
area_mm2, gauge_length_mm = specimen_geometry

# Receive tuple return values from function.
result_tuple = summarize_tensile_test(tensile_loads_kn, area_mm2)

# Unpack tuple return positions clearly.
ultimate_mpa, first_mpa, status = result_tuple

# Print concise teaching results.
print("Returned tuple:", result_tuple)
print("Ultimate strength MPa:", round(ultimate_mpa, 1))
print("First stress MPa:", round(first_mpa, 1))
print("Diagnostic status:", status)
print("Tuple length stays fixed:", len(result_tuple))

```
