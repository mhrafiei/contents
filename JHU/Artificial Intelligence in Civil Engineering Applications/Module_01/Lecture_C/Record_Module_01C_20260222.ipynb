### **1.2. Lists and Dictionaries**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_01_02.jpg?v=1771809445" width="250">



>* Use lists for ordered, similar engineering values.
>* Use dictionaries for labeled lookups and quick access.

>* Combine lists and dictionaries for structured datasets
>* Use keys, avoid fragile index positions

>* Use lists to clean and align sequences.
>* Use dictionaries for missing fields, safer summaries.



```python
#@title Python Code - Lists and Dictionaries

# Practice lists and dictionaries with civil engineering data.
# Use loops and conditions to clean small datasets.
# Print short summaries and handle missing values safely.

# Store daily rainfall totals in millimeters.
rain_mm = [0.0, 12.5, None, 8.0, -1.0, 25.0, 300.0]
# Define plausible rainfall bounds for quick cleaning.
min_ok, max_ok = 0.0, 200.0

# Build a cleaned list while preserving original order.
clean_rain = []
# Track how many values were removed as invalid.
removed_count = 0

# Loop through each raw measurement and validate it.
for value in rain_mm:
# Skip missing entries represented by None.
    if value is None:
        removed_count = removed_count + 1
        continue

# Skip values outside a plausible engineering range.
    if value < min_ok or value > max_ok:
        removed_count = removed_count + 1
        continue

# Keep valid values in a new list.
    clean_rain.append(value)

# Compute a simple average if any values remain.
if len(clean_rain) > 0:
    avg_rain = sum(clean_rain) / len(clean_rain)
else:
    avg_rain = 0.0

# Print a short list summary for quick debugging.
print("Valid rainfall values:", clean_rain)
print("Removed invalid entries:", removed_count)
print("Average valid rainfall:", round(avg_rain, 2))

# Map sensor identifiers to their latest water level.
sensor_levels_m = {"PZ-01": 1.25, "PZ-02": None, "PZ-03": 0.95}
# Provide a safe default when a sensor is missing.
default_level = 0.0

# Update only sensors that reported a valid reading.
for sensor_id in sensor_levels_m:
# Read the current value using the dictionary key.
    level = sensor_levels_m[sensor_id]

# Replace missing readings with a safe default.
    if level is None:
        sensor_levels_m[sensor_id] = default_level

# Add a new sensor record using key based access.
sensor_levels_m["PZ-04"] = 1.10
# Print a compact dictionary summary.
print("Sensor levels now:", sensor_levels_m)

# Combine list and dictionaries for inspection observations.
inspections = [
    {"id": "B-101", "state": 2, "defect": "crack"},
    {"id": "B-102", "state": 4, "defect": "corrosion"},
    {"id": "B-103", "state": 1, "defect": "none"},
]

# Count how many elements need attention using control flow.
needs_attention = 0
# Use try except to handle unexpected missing keys.
try:
    for obs in inspections:
        if obs["state"] >= 3:
            needs_attention = needs_attention + 1
except KeyError as err:
    print("Missing key found:", str(err))

# Print a final short result for the inspection list.
print("Elements needing attention:", needs_attention)

```

### **1.3. Control Flow Loops**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_01_03.jpg?v=1771809479" width="250">



>* Repeat steps across many engineering measurements
>* Automate consistent cleaning, avoiding manual errors

>* Use loops with if-statements to handle exceptions.
>* Track quality counts and stop early.

>* Use loops to build totals, stats, groups
>* Update dictionaries carefully for reproducible summaries



```python
#@title Python Code - Control Flow Loops

# Learn loops using small civil engineering measurements.
# Practice condition checks while iterating through records.
# Summarize cleaned results for quick quality control.

# Create rainfall intensities with a few bad entries.
rain_mm_per_hr = [12.0, 0.0, -3.0, None, 25.5, 8.2]
# Set a conversion factor for unit conversion.
mm_to_in = 1.0 / 25.4

# Prepare containers for cleaned values and flags.
clean_in_per_hr = []
flagged_notes = []
# Initialize counters for accepted and rejected records.
accepted_count, rejected_count = 0, 0

# Loop through each measurement and apply checks.
for idx, value in enumerate(rain_mm_per_hr):
    # Handle missing values without stopping the script.
    if value is None:
        rejected_count += 1
        flagged_notes.append((idx, "missing"))

    # Handle impossible negative intensities as rejected.
    elif value < 0:
        rejected_count += 1
        flagged_notes.append((idx, "negative"))

    # Convert valid values and store the cleaned result.
    else:
        accepted_count += 1
        clean_in_per_hr.append(value * mm_to_in)

# Compute a simple summary using a loop.
total_in = 0.0
# Accumulate totals to avoid advanced functions.
for x in clean_in_per_hr:
    total_in += x

# Safely compute an average with a conditional.
if len(clean_in_per_hr) > 0:
    avg_in = total_in / len(clean_in_per_hr)
else:
    avg_in = 0.0

# Print a compact quality control summary.
print("Accepted records:", accepted_count)
print("Rejected records:", rejected_count)
print("Average intensity (in/hr):", round(avg_in, 3))
print("Flagged entries:", flagged_notes)

```

## **2. Reusable Python Functions**

### **2.1. Simple Functions and Classes**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_02_01.jpg?v=1771809505" width="250">



>* Turn repeated steps into reusable functions.
>* Centralize cleaning logic for consistent, readable scripts.

>* Use classes for related tasks sharing settings.
>* Model engineering objects to avoid messy scripts.

>* Centralize logic to handle changing datasets easily.
>* Simplifies testing, debugging, and reuse across analyses.



```python
#@title Python Code - Simple Functions and Classes

# Reuse functions to clean small engineering datasets.
# Build simple classes for repeated project processing.
# Add error handling for safer calculations.

import math
import random

# Set a deterministic seed for repeatable results.
random.seed(7)

# Create a tiny rainfall dataset with messy headers.
raw_records = [
    {"Station ID": "A1", "Rain(mm)": 12, "Day": "Mon"},
    {"Station ID": "A1", "Rain(mm)": -3, "Day": "Tue"},
    {"Station ID": "B2", "Rain(mm)": 5, "Day": "Mon"},
    {"Station ID": "B2", "Rain(mm)": None, "Day": "Tue"},
]

# Standardize keys into consistent snake_case names.
def standardize_keys(record):
    # Build a new dictionary with cleaned key names.
    cleaned = {}
    # Loop through each key value pair.
    for key, value in record.items():
        new_key = key.strip().lower().replace(" ", "_")
        cleaned[new_key] = value
    return cleaned

# Convert millimeters to meters with basic validation.
def mm_to_m(value_mm):
    # Reject missing values early for safer math.
    if value_mm is None:
        raise ValueError("Missing rainfall value")
    # Reject impossible negative rainfall measurements.
    if value_mm < 0:
        raise ValueError("Negative rainfall is invalid")
    return value_mm / 1000.0

# Summarize rainfall by station using reusable functions.
def summarize_rainfall_m(records):
    # Accumulate totals in a dictionary by station.
    totals = {}
    # Track how many records were skipped.
    skipped = 0
    # Process each record with defensive error handling.
    for rec in records:
        clean = standardize_keys(rec)
        station = clean.get("station_id", "unknown")
        try:
            rain_m = mm_to_m(clean.get("rain(mm)"))
            totals[station] = totals.get(station, 0.0) + rain_m
        except ValueError:
            skipped = skipped + 1
    return totals, skipped

# Create a class to bundle settings and behaviors.
class ProjectRainProcessor:
    # Store project settings like maximum plausible rainfall.
    def __init__(self, project_name, max_mm):
        self.project_name = project_name
        self.max_mm = max_mm

    # Validate and convert rainfall using project rules.
    def safe_rain_m(self, value_mm):
        # Reuse the shared conversion function first.
        rain_m = mm_to_m(value_mm)
        # Apply a project specific upper bound check.
        if value_mm > self.max_mm:
            raise ValueError("Rainfall exceeds project maximum")
        return rain_m

    # Process records and return totals plus flags.
    def process(self, records):
        # Accumulate totals and count flagged records.
        totals = {}
        flagged = 0
        # Loop through records and apply class behavior.
        for rec in records:
            clean = standardize_keys(rec)
            station = clean.get("station_id", "unknown")
            try:
                rain_m = self.safe_rain_m(clean.get("rain(mm)"))
                totals[station] = totals.get(station, 0.0) + rain_m
            except ValueError:
                flagged = flagged + 1
        return totals, flagged

# Run the function based workflow on the raw records.
func_totals, func_skipped = summarize_rainfall_m(raw_records)

# Run the class based workflow with a project rule.
processor = ProjectRainProcessor("Drainage Study", max_mm=50)
class_totals, class_flagged = processor.process(raw_records)

# Print compact results without dumping large structures.
print("Function totals (m):", func_totals)
print("Function skipped records:", func_skipped)
print("Class totals (m):", class_totals)
print("Class flagged records:", class_flagged)

# Show how a single bad value triggers safe debugging.
try:
    demo_value = processor.safe_rain_m(120)
    print("Converted demo rainfall (m):", demo_value)
except ValueError as err:
    print("Debug message:", str(err))

```

### **2.2. Arguments Methods Returns**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_02_02.jpg?v=1771809543" width="250">



>* Choose arguments that match real data decisions
>* Return consistent outputs others can rely on

>* Methods use class data, fewer arguments needed.
>* Returns versus updates affect testing, reuse.

>* Use clear, consistent arguments and input patterns.
>* Return consistent results plus change logs, metadata.



```python
#@title Python Code - Arguments Methods Returns

# Learn function arguments, methods, and return values.
# Use small civil datasets for reusable calculations.
# Practice debugging with simple error handling.

# Import only built-in tools for this recap.
import math

# Define a reusable unit conversion function.
def convert_lengths(values, from_unit, to_unit):

    # Validate arguments and raise clear errors.
    if values is None or len(values) == 0:
        raise ValueError("values must be a non-empty list")

    # Store supported conversion factors in meters.
    factors = {"mm": 0.001, "m": 1.0, "ft": 0.3048}

    # Check that requested units are supported.
    if from_unit not in factors or to_unit not in factors:
        raise ValueError("Unsupported unit; use mm, m, or ft")

    # Convert each value using a simple loop.
    converted = []

    # Loop through values and validate each entry.
    for x in values:
        if x is None or x <= 0:
            raise ValueError("All lengths must be positive numbers")

        # Convert via meters, then to target unit.
        meters = x * factors[from_unit]
        converted.append(meters / factors[to_unit])

    # Return a predictable output type.
    return converted

# Define a reusable cleaning function with thresholds.
def clean_sensor_readings(readings, min_ok, max_ok):

    # Validate the threshold arguments.
    if min_ok >= max_ok:
        raise ValueError("min_ok must be smaller than max_ok")

    # Prepare outputs for cleaned data and report.
    cleaned = []

    # Track how many readings were removed.
    removed = 0

    # Loop through readings and filter invalid values.
    for r in readings:
        if r is None:
            removed = removed + 1

        elif r < min_ok or r > max_ok:
            removed = removed + 1

        else:
            cleaned.append(r)

    # Return both cleaned data and a small report.
    report = {"kept": len(cleaned), "removed": removed}

    # Return a tuple so callers can unpack results.
    return cleaned, report

# Create a class to demonstrate methods and state.
class LabSeries:

    # Initialize stored readings and unit metadata.
    def __init__(self, readings, unit):
        self.readings = list(readings)

        # Store unit as a simple string label.
        self.unit = unit

    # Method cleans internal readings and returns report.
    def clean_in_place(self, min_ok, max_ok):

        # Reuse the function and update object state.
        cleaned, report = clean_sensor_readings(
            self.readings, min_ok, max_ok
        )

        # Update stored readings for later methods.
        self.readings = cleaned

        # Return a small report for traceability.
        return report

    # Method computes a summary and returns a dictionary.
    def mean_value(self):

        # Guard against empty data after cleaning.
        if len(self.readings) == 0:
            return {"mean": None, "unit": self.unit}

        # Compute mean using a simple loop.
        total = 0.0

        # Sum values without printing large outputs.
        for r in self.readings:
            total = total + r

        # Return consistent keys for downstream code.
        return {"mean": total / len(self.readings), "unit": self.unit}

# Create a tiny civil engineering example dataset.
raw_lengths_ft = [10.0, 12.5, 9.0]

# Create a tiny sensor dataset with invalid entries.
raw_pressures_kpa = [101.2, None, 98.7, 250.0, 100.1]

# Demonstrate arguments and returns with try handling.
try:
    lengths_m = convert_lengths(raw_lengths_ft, "ft", "m")

    # Print a short, readable result summary.
    print("Converted lengths (m):", [round(x, 3) for x in lengths_m])

except ValueError as err:
    print("Conversion error:", err)

# Demonstrate cleaning function returns two outputs.
try:
    cleaned_p, report_p = clean_sensor_readings(
        raw_pressures_kpa, 90.0, 120.0
    )

    # Print only small summaries, not full arrays.
    print("Cleaning report:", report_p)

except ValueError as err:
    print("Cleaning error:", err)

# Demonstrate methods that operate on stored state.
series = LabSeries(raw_pressures_kpa, "kPa")

# Clean using a method, then compute a mean.
report2 = series.clean_in_place(90.0, 120.0)

# Print method return values for debugging clarity.
print("Method report:", report2)

# Print a final summary dictionary from a method.
print("Mean summary:", series.mean_value())
```

## **3. Lightweight Debugging Basics**

### **3.1. Print Tracing**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_03_01.jpg?v=1771809620" width="250">



>* Print checkpoints reveal what your script does.
>* Print key shapes, samples, and progress.

>* Print to answer specific debugging questions.
>* Use breadcrumbs to find first mismatch.

>* Use clear, labeled, summary prints; remove later.
>* Add progress prints at key steps.



```python
#@title Python Code - Print Tracing

# Print tracing reveals hidden program state quickly.
# We debug a tiny civil dataset stepwise.
# Each print answers one focused question.

# Create a small station dataset with issues.
stations = [
    {"id": " ST-01 ", "slope_pct": "2.5"},
    {"id": "st-02", "slope_pct": "-1.0"},
    {"id": "ST-03", "slope_pct": "N/A"},
]

# Print a checkpoint about loaded records.
print("Loaded records:", len(stations))

# Define a cleaning function with tracing.
def clean_station_id(raw_id):
    # Print the raw identifier before cleaning.
    print("Raw id:", raw_id)
    cleaned = raw_id.strip().upper()
    # Print the cleaned identifier after cleaning.
    print("Cleaned id:", cleaned)
    return cleaned

# Define a safe parser with error handling.
def parse_slope_percent(value_text):
    # Print the raw slope text before parsing.
    print("Slope text:", value_text)
    try:
        slope = float(value_text)
    except ValueError:
        # Print a short warning for bad values.
        print("Bad slope, using None:", value_text)
        slope = None
    return slope

# Process records and trace key decisions.
cleaned = []
for rec in stations:
    # Print which record is being processed.
    print("Processing record id:", rec["id"])
    sid = clean_station_id(rec["id"])
    slope = parse_slope_percent(rec["slope_pct"])

    # Print a decision checkpoint for filtering.
    if slope is None or slope <= 0:
        print("Skipping record:", sid)
        continue

    # Print confirmation when a record is kept.
    print("Keeping record:", sid)
    cleaned.append({"id": sid, "slope_pct": slope})

# Print a final summary checkpoint.
print("Kept records:", len(cleaned))

```

### **3.2. Common Exception Handling**

<img src="https://cdn.jsdelivr.net/gh/mhrafiei/contents@main/JHU/Artificial Intelligence in Civil Engineering Applications/Module_01/Lecture_C/image_03_02.jpg?v=1771809653" width="250">



>* Exceptions reveal wrong assumptions in scripts.
>* Catch expected errors; print context; stop/fallback.

>* Anticipate file, conversion, key, and index errors.
>* Print context, bad input, and next checks.

>* Choose stop or continue; log failures.
>* Catch specific errors; avoid silent broad catches.



```python
#@title Python Code - Common Exception Handling

# This script shows common exceptions in civil datasets.
# We catch errors and print clear debugging messages.
# Examples use small lists and dictionaries only.

# Create a tiny survey dataset with one bad value.
survey_points = [
    {"Station": "0+000", "Elev_m": "12.4"},
    {"Station": "0+010", "Elev_m": "NA"},
    {"Station": "0+020", "Elev_m": "13.1"},
]

# Convert elevation text to float, with safe fallback.
def safe_float(text_value, fallback_value):
    try:
        return float(text_value)
    except ValueError:
        print("ValueError: cannot convert", text_value, "using", fallback_value)
        return fallback_value

# Compute average elevation, skipping missing values.
def average_elevation(points_list):
    total_sum, count_ok = 0.0, 0
    for point in points_list:
        elev = safe_float(point["Elev_m"], None)
        if elev is not None:
            total_sum, count_ok = total_sum + elev, count_ok + 1
    if count_ok == 0:
        raise ZeroDivisionError("No valid elevations for averaging")
    return total_sum / count_ok

# Demonstrate KeyError when a column name changes.
def get_station_label(point_dict):
    try:
        return point_dict["Station"]
    except KeyError:
        print("KeyError: missing Station key, available:", list(point_dict.keys()))
        return "UNKNOWN"

# Demonstrate IndexError when a list is shorter than expected.
def get_point_by_index(points_list, index_value):
    try:
        return points_list[index_value]
    except IndexError:
        print("IndexError: index", index_value, "out of range, length", len(points_list))
        return None

# Demonstrate FileNotFoundError for a missing input file.
def read_first_line(file_path):
    try:
        f = open(file_path, "r")
        first = f.readline().strip()
        f.close()
        return first
    except FileNotFoundError:
        print("FileNotFoundError: check path and filename:", file_path)
        return None

# Run the examples and print short, useful results.
avg = average_elevation(survey_points)
print("Average elevation (m):", round(avg, 2))

# Trigger a KeyError by using a renamed key.
bad_point = {"STA": "0+030", "Elev_m": "14.0"}
print("Station label:", get_station_label(bad_point))

# Trigger an IndexError by requesting a missing row.
missing = get_point_by_index(survey_points, 10)
print("Missing point returned:", missing)

# Trigger a FileNotFoundError with a fake path.
line = read_first_line("./datasets/does_not_exist.csv")
print("First line read:", line)

```
